## Lecture Notes

### Overview

- A **torsion pendulum** is the angular analog of a spring–block oscillator: a disk or object suspended by a wire that twists.
- **Springiness** comes from the wire's resistance to twisting; **inertia** comes from the object's rotational inertia $I$.
- The same SHM structure applies: restoring torque $\tau \propto -\theta$, with period $T = 2\pi\sqrt{I/\kappa}$.

| Linear | Angular |
|--------|---------|
| Displacement $x$ | Angular displacement $\theta$ |
| Force $F = -kx$ | Torque $\tau = -\kappa\theta$ |
| Mass $m$ | Rotational inertia $I$ |
| Spring constant $k$ | Torsion constant $\kappa$ |
| $T = 2\pi\sqrt{m/k}$ | $T = 2\pi\sqrt{I/\kappa}$ |

---

### The torsion pendulum

A **torsion pendulum** consists of an object (e.g., a disk) suspended from a wire. When the wire is twisted and released, the object oscillates in angular simple harmonic motion.

- **Restoring torque**: twisting the wire produces a torque that opposes the displacement:

$$
\tau = -\kappa\theta
$$ (eq-torsion-torque)

- **κ (kappa)** is the **torsion constant**—a property of the wire (length, diameter, material). It plays the analog of the spring constant $k$.

:::{admonition} Angular form of Hooke's law
:class: note

Just as $F = -kx$ gives restoring force proportional to displacement, $\tau = -\kappa\theta$ gives restoring torque proportional to angular displacement. Both are linear (first power) restoring laws.
:::

---

### Period and angular frequency

From $\tau = I\alpha$ (Newton's second law for rotation) and $\tau = -\kappa\theta$:

$$
I\alpha = -\kappa\theta \quad \Rightarrow \quad \alpha = -\frac{\kappa}{I} \theta
$$ (eq-alpha-theta)

Comparing with the hallmark of SHM, $\alpha = -\omega^2\theta$, we identify:

$$
\omega = \sqrt{\frac{\kappa}{I}}
$$ (eq-omega-torsion)

$$
T = 2\pi\sqrt{\frac{I}{\kappa}}
$$ (eq-period-torsion)

- **Larger inertia** $I$ → longer period (slower oscillations).
- **Stiffer wire** (larger $\kappa$) → shorter period (faster oscillations).

:::{important}
The torsion constant $\kappa$ depends only on the wire. If two different objects hang from the **same wire**, $\kappa$ is the same; only $I$ differs. So $T \propto \sqrt{I}$—measuring periods lets you compare rotational inertias.
:::

---

### Comparing two torsion pendulums

Same wire (same $\kappa$), different objects $a$ and $b$:

$$
\frac{T_b}{T_a} = \sqrt{\frac{I_b}{I_a}} \quad \Rightarrow \quad I_b = I_a \left(\frac{T_b}{T_a}\right)^2
$$ (eq-compare-I)

If $T_b > T_a$, then $I_b > I_a$—the object with the longer period has the larger rotational inertia.

### Summary

- **Torsion pendulum**: object on a twisted wire; restoring torque $\tau = -\kappa\theta$ (angular Hooke's law).
- **Torsion constant** $\kappa$: property of the wire; plays the role of spring constant $k$.
- **Period** $T = 2\pi\sqrt{I/\kappa}$; **angular frequency** $\omega = \sqrt{\kappa/I}$.
- **Same wire, different objects**: $I_b/I_a = (T_b/T_a)^2$—measure periods to compare rotational inertias.
