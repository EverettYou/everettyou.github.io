## Lecture Notes

### Overview

- **Pascal's principle**: A change in pressure applied to an **enclosed incompressible fluid** is transmitted **undiminished** to every part of the fluid and to the container walls.
- **Application**: Hydraulic lifts, brakes, jacks—small input force can produce large output force.
- **Trade-off**: Force amplification comes at the cost of displacement; work is conserved.

---

### Pascal's principle

**Statement** (Blaise Pascal, 1652): A change in the pressure applied to an enclosed incompressible fluid is transmitted undiminished to every portion of the fluid and to the walls of its container.

**Why incompressible?** If the fluid compressed, the pressure change would not propagate uniformly. Liquids are nearly incompressible; the principle applies well to them.

**Demonstration**: Add lead shot to a piston atop a liquid-filled cylinder. The extra pressure $\Delta p_{\text{ext}}$ appears equally at every point in the liquid: $\Delta p = \Delta p_{\text{ext}}$, independent of depth.

**Examples**: Toothpaste tube (squeeze one end → pressure transmitted → paste exits); Heimlich maneuver (pressure on abdomen → transmitted to throat); hydraulic brakes.

---

### Hydraulic lever (hydraulic lift)

A **hydraulic lever** has two pistons of different areas $A_i$ (input) and $A_o$ (output), connected by an incompressible liquid. Same pressure $p$ throughout:

$$
\frac{F_i}{A_i} = p = \frac{F_o}{A_o}
$$ (eq-hydraulic-pressure)

**Force amplification**:
$$
\frac{F_o}{F_i} = \frac{A_o}{A_i}
$$ (eq-hydraulic-force)

- Small input force $F_i$ on small area → large output force $F_o$ on large area
- The **ratio of forces** equals the **ratio of areas**

**Displacement trade-off**: The volume of liquid displaced by the input piston equals that received by the output piston:

$$
A_i d_i = A_o d_o
$$ (eq-hydraulic-volume)

- $d_i$: input piston displacement (down)
- $d_o$: output piston displacement (up)

So
$$
\frac{d_o}{d_i} = \frac{A_i}{A_o}
$$ (eq-hydraulic-displacement)

The output piston moves **less** when $A_o > A_i$. You push the small piston through a large distance to lift the load a small distance.

:::{important}
**Work is conserved**: $W_i = F_i d_i = F_o d_o = W_o$. The hydraulic lever does not create energy; it trades force for distance (or vice versa).
:::

### Summary

- **Pascal's principle**: Pressure change in enclosed incompressible fluid transmitted undiminished everywhere
- **Hydraulic force**: $F_o/F_i = A_o/A_i$
- **Hydraulic displacement**: $A_i d_i = A_o d_o$; $d_o/d_i = A_i/A_o$
- **Work**: $F_i d_i = F_o d_o$ (no free lunch)
