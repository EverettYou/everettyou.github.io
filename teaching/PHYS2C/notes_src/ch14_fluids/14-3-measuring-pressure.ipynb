# 14-3 Measuring Pressure


## Learning Objectives

*After reading this module, you should be able to . . .*

- **14.06** Describe how a barometer can measure atmospheric pressure.
- **14.07** Describe how an open-tube manometer can measure the gauge pressure of a gas.

## Key Ideas

A mercury barometer can be used to measure atmospheric pressure.

An open-tube manometer can be used to measure the gauge pressure of a confined gas.
