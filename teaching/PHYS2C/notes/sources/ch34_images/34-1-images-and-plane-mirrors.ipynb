## Key Ideas 

- An image is a reproduction of an object via light. If the 

- image can form on a surface, it is a real image and can exist even if no observer is present. If the image requires the visual system of an observer, it is a virtual image. 

- A plane (flat) mirror can form a virtual image of a light 

- source (said to be the object) by redirecting light rays emerging from the source. The image can be seen where backward
