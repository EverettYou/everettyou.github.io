# 14-2 Fluids at Rest


## Learning Objectives

- **14.05** Distinguish between total pressure (absolute pressure) and gauge pressure.

## Key Ideas

Pressure in a fluid at rest varies with vertical position $y$. For $y$ measured positive upward,

$$p_2 - p_1 = \rho g (y_1 - y_2)$$

If $h$ is the *depth* of a fluid sample *below* some reference level at which the pressure is $p_0$, this equation becomes

$$p = p_0 + \rho g h$$

where $p$ is the pressure in the sample.

The pressure in a fluid is the same for all points at the same level.

Gauge pressure is the difference between the actual pressure (or absolute pressure) at a point and the atmospheric pressure.
