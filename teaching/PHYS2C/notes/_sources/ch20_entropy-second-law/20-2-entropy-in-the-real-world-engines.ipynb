## Key Ideas 

● An irreversible process is one that cannot be reversed by means of small changes in the environment. The direction in which an irreversible process proceeds is set by the change in entropy � _S_ of the system undergoing the process. Entropy _S_ is a state property (or state function) of the system; that is, it depends only on the state of the system and not on the way in which the system reached that state. The entropy postulate states (in part): If an irreversible process occurs in a closed system, the entropy of the system always increases.
