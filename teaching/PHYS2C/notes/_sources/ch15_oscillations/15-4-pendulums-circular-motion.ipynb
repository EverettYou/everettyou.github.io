## Lecture Notes

### Overview

- **Pendulums** are SHM oscillators where **gravity** provides the restoring torque (not a spring or twisted wire).
- **Simple pendulum**: point mass on a string. **Physical pendulum**: extended body; period depends on $I$ and distance $h$ from pivot to center of mass.
- **SHM = projection of uniform circular motion** onto a diameter—a deep connection that explains why $x$, $v$, and $a$ are sinusoidal.

| Type | Restoring mechanism | Period |
|------|---------------------|--------|
| Spring–block | $F = -kx$ | $T = 2\pi\sqrt{m/k}$ |
| Torsion | $\tau = -\kappa\theta$ | $T = 2\pi\sqrt{I/\kappa}$ |
| Simple pendulum | Gravity ($\tau \propto -\theta$) | $T = 2\pi\sqrt{L/g}$ |
| Physical pendulum | Gravity ($\tau \propto -\theta$) | $T = 2\pi\sqrt{I/(mgh)}$ |

---

### The simple pendulum

A **simple pendulum** is a particle (bob) of mass $m$ on a massless string of length $L$, pivoted at the top.

- **Restoring torque**: The tangential component of gravity, $mg\sin\theta$, produces a torque about the pivot. For small $\theta$, $\sin\theta \approx \theta$ (in radians):

$$
\tau = -L(mg\sin\theta) \approx -mgL\theta
$$ (eq-simple-torque)

- **Angular equation**: $\tau = I\alpha$ with $I = mL^2$ gives $\alpha = -(g/L)\theta$—the hallmark of SHM.

$$
\omega = \sqrt{\frac{g}{L}}, \qquad T = 2\pi\sqrt{\frac{L}{g}}
$$ (eq-simple-period)

- **Independent of mass**: $T$ depends only on $L$ and $g$. Heavier and lighter bobs swing at the same rate.

:::{caution}
**Small-angle only.** The approximation $\sin\theta \approx \theta$ holds for $\theta \lesssim 10°$. For larger amplitudes, the motion is periodic but not sinusoidal; $T$ increases slightly.
:::

---

### The physical pendulum

A **physical pendulum** is any rigid body that swings about a pivot (not at its center of mass).

- **Restoring torque**: Gravity acts at the center of mass, distance $h$ from the pivot. The moment arm of $mg\sin\theta$ is $h$:

$$
\tau = -h(mg\sin\theta) \approx -mgh\theta
$$ (eq-physical-torque)

- **Period** (small angles):

$$
T = 2\pi\sqrt{\frac{I}{mgh}}
$$ (eq-physical-period)

- $I$ = rotational inertia about the pivot; $h$ = distance from pivot to center of mass.
- If the pivot is at the center of mass, $h = 0$ and $T \to \infty$—no oscillation.

:::{admonition} Simple vs physical
:class: note

A **simple pendulum** is a special case: all mass at distance $L$, so $I = mL^2$ and $h = L$. Then [](#eq-physical-period) reduces to [](#eq-simple-period).
:::

---

### Center of oscillation

For any physical pendulum with period $T$, there is an **equivalent simple pendulum** of length $L_0$ with the same period:

$$
T = 2\pi\sqrt{\frac{L_0}{g}} = 2\pi\sqrt{\frac{I}{mgh}} \quad \Rightarrow \quad L_0 = \frac{I}{mh}
$$ (eq-L0)

The point at distance $L_0$ from the pivot is the **center of oscillation**. For a uniform rod pivoted at one end: $I = \frac{1}{3}mL^2$, $h = L/2$, so $L_0 = \frac{2}{3}L$.

---

### SHM and uniform circular motion

**Simple harmonic motion is the projection of uniform circular motion onto a diameter.**

- Imagine a particle $P'$ moving in a circle of radius $x_m$ at constant angular speed $\omega$.
- Its **projection** $P$ onto the $x$-axis executes SHM.
- At time $t$, the angle is $\omega t + \phi$; the $x$-component of the position vector is $x_m\cos(\omega t + \phi)$—exactly [](#eq-shm-displacement) from section 15-1.
- Similarly, the projections of the velocity and (radial) acceleration vectors give $v(t)$ and $a(t)$ for SHM.

:::{important}
This explains why SHM is sinusoidal: it is "circular motion viewed edge-on." The phase $\omega t + \phi$ is the actual angle of the reference particle on the circle.
:::

---

### Summary

- **Simple pendulum**: $T = 2\pi\sqrt{L/g}$; independent of mass; valid for small angles.
- **Physical pendulum**: $T = 2\pi\sqrt{I/(mgh)}$; $h$ = pivot-to-CM distance.
- **Center of oscillation**: $L_0 = I/(mh)$—equivalent simple-pendulum length.
- **SHM = projection of uniform circular motion** onto a diameter; the phase is the angle on the reference circle.
