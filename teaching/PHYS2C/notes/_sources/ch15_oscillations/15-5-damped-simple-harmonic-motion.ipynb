## Lecture Notes

### Overview: Damped SHM

- **Real oscillators** lose energy to friction, air resistance, or other dissipative forces. The motion is **damped**—amplitude and energy decrease over time.
- **Damping force** $F_d = -bv$: proportional to velocity, opposes motion. Energy is transferred to thermal energy.
- **Displacement**: $x(t) = x_m e^{-bt/2m}\cos(\omega' t + \phi)$—a cosine with an **exponentially decaying amplitude**.

| Undamped | Damped |
|----------|--------|
| Amplitude constant | Amplitude $\propto e^{-bt/2m}$ |
| $E$ constant | $E \propto e^{-bt/m}$ |
| $\omega = \sqrt{k/m}$ | $\omega' = \sqrt{k/m - b^2/(4m^2)}$ |

---

### The damping force

A common model: the damping force is proportional to velocity and opposes it:

$$
F_d = -bv
$$ (eq-damping-force)

- **$b$** = damping constant (SI: kg/s). Depends on the vane, fluid, or friction mechanism.
- The minus sign ensures $F_d$ opposes the direction of motion.

Newton's second law: $-bv - kx = ma$, giving

$$
m\frac{d^2x}{dt^2} + b\frac{dx}{dt} + kx = 0
$$ (eq-damped-de)

---

### Displacement and damped frequency

The solution (for $b < 2\sqrt{km}$, underdamped case) is

$$
x(t) = x_m e^{-bt/2m} \cos(\omega' t + \phi)
$$ (eq-damped-x)

- **Amplitude envelope**: $x_m e^{-bt/2m}$—decays exponentially.
- **Damped angular frequency**:

$$
\omega' = \sqrt{\frac{k}{m} - \frac{b^2}{4m^2}}
$$ (eq-omega-prime)

- If $b = 0$: $\omega' = \omega = \sqrt{k/m}$ (undamped).
- If $b \ll \sqrt{km}$ (light damping): $\omega' \approx \omega$—period nearly unchanged; only amplitude decays.

:::{admonition} Overdamping and critical damping
:class: seealso

If $b \geq 2\sqrt{km}$, the motion is **overdamped** (no oscillation) or **critically damped** (fastest return to equilibrium without overshoot). The cosine form [](#eq-damped-x) applies only when $b < 2\sqrt{km}$.
:::

---

### Energy decay

Mechanical energy $E = K + U$ is **not** conserved—it decreases as work is done against the damping force.

For **small damping**, the amplitude is $x_m e^{-bt/2m}$, so (replacing the constant amplitude in $E = \frac{1}{2}kx_m^2$):

$$
E(t) \approx \frac{1}{2}kx_m^2\, e^{-bt/m}
$$ (eq-damped-energy)

- **Time constant** for energy decay: $\tau = m/b$. After time $\tau$, $E$ drops by factor $e^{-1} \approx 0.37$.
- **Time for $E$ to drop to fraction $f$** of initial: $e^{-bt/m} = f$ $\Rightarrow$ $t = \frac{m}{b}\ln(1/f)$.

### Summary

- **Damping**: $F_d = -bv$; energy lost to thermal.
- **$x(t) = x_m e^{-bt/2m}\cos(\omega' t + \phi)$**; amplitude decays exponentially.
- **$\omega' = \sqrt{k/m - b^2/(4m^2)}$**; for small $b$, $\omega' \approx \omega$.
- **$E(t) \approx \frac{1}{2}kx_m^2\, e^{-bt/m}$**; time constant $\tau = m/b$.
