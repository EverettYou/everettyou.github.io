## Lecture Notes

### Overview

This lesson unifies two classical frameworks: geometric optics (ray picture) and classical mechanics (particle picture). Both obey the same variational principle—Fermat's principle for light, Hamilton's principle for particles. The bridge between them is **de Broglie's insight**: matter has waves, and the phase of the matter wave *is the classical action*.

### Optical Path as Action

- **Geometric optics**: light follows the path of *stationary optical path length*: $L = \int n \, \mathrm{d}s$.

- **Classical mechanics**: a particle follows the path of *stationary action*: $S = \int L \, \mathrm{d}t$.

These have *identical structure*:

| **Geometric Optics** | **Classical Mechanics** |
|---|---|
| Light ray | Particle trajectory |
| Fermat's principle: $\delta L = 0$ | Hamilton's principle: $\delta S = 0$ |
| Optical path $L = \int n \, \mathrm{d}s$ | Action $S = \int L \, \mathrm{d}t$ |
| Ray optics (short wavelength) | Classical limit ($\hbar \to 0$) |

The physical picture: both light and particles are described by a variational principle. The observable path is the one that makes the optical path length (or action) stationary.

### De Broglie: Matter Waves

In 1924, de Broglie proposed that matter, like light, has **wave-particle duality**.

:::{admonition} De Broglie Relations
:class: important

$$
\boldsymbol{p} = \hbar \boldsymbol{k}, \quad E = \hbar \omega
$$

which associate: 
- **particle** properties: momentum $\boldsymbol{p}$ and energy $E$, with
- **wave** properties: wave vector $\boldsymbol{k}$ and angular frequency $\omega$.

:::

This unifies particle and wave descriptions. **Experimental confirmation** (Davisson & Germer, 1927): electrons diffracted by crystals exhibit interference patterns with $\lambda = h/p$, confirming de Broglie's hypothesis.

:::{admonition} Derivation: De Broglie Relations from the Optics–Mechanics Analogy
:class: dropdown information

The key reasoning is by analogy with light.

**For light (photons):** Einstein (1905) and Planck (1900) established $E = \hbar\omega$ and $p = \hbar k = h/\lambda$ for photons. These relate particle properties (energy, momentum) to wave properties (frequency, wave vector).

**De Broglie's hypothesis:** If light (a wave) can behave as particles, then matter (particles) should behave as waves. De Broglie proposed the *same* relations hold for massive particles:

$$
E = \hbar\omega \quad \Longleftrightarrow \quad \omega = E/\hbar
$$

$$
\boldsymbol{p} = \hbar \boldsymbol{k} \quad \Longleftrightarrow \quad \lambda = h/p
$$

**Consistency check with optics.** For light in a medium of refractive index $n$, the local wave number is $k_{\text{loc}} = n k = n\omega/c$. The optical momentum is $p_{\text{opt}} = n \cdot (E/c) = \hbar k_{\text{loc}}$. So the de Broglie relation $p = \hbar k$ is already implicit in geometric optics — de Broglie simply extended it to massive particles.

**For a free non-relativistic particle** with $E = p^2/2m$, the de Broglie wavelength is

$$
\lambda = \frac{h}{p} = \frac{h}{\sqrt{2mE}}
$$

For an electron at $E = 100\,\text{eV}$: $\lambda \approx 0.12\,\text{nm}$, comparable to atomic spacings — hence crystal diffraction is observable.
:::

### Action = Phase: The Punchline

Here is the crucial insight connecting classical and quantum mechanics.

In **optics**, the phase accumulated by a wave traveling through a medium is:

$$
\phi_\text{optics} = \frac{2\pi}{\lambda} \int n \, \mathrm{d}s
$$

In **classical mechanics**, the accumulated action along a path is:

$$
S = \int L \, \mathrm{d}t
$$

**De Broglie's identification**: The phase of the quantum wavefunction is the classical action in units of $\hbar$:

$$
\Phi = \frac{S}{\hbar}
$$

Therefore, the quantum wavefunction is:

$$
\boxed{\psi(\boldsymbol{x}, t) = A(\boldsymbol{x}, t) \, \mathrm{e}^{\mathrm{i}S(\boldsymbol{x}, t)/\hbar}}
$$

**This is the bridge.** Phase is action. Once we know the action for any physical system, we can immediately write down the phase of its quantum wavefunction.

:::{admonition} Derivation: Action = Phase from the Optics–Mechanics Dictionary
:class: dropdown information

The proof proceeds by matching the optics and mechanics columns of the correspondence table.

**In optics**, a monochromatic wave in a medium with $n(\boldsymbol{r})$ has the form

$$
\psi_{\text{opt}} = A \, \mathrm{e}^{\mathrm{i}\Phi}, \quad \Phi = k \int n \, \mathrm{d}s = \frac{\omega}{c} \int n \, \mathrm{d}s = \frac{\omega}{c} L
$$

where $L$ is the optical path length.

**In mechanics**, the action along a classical trajectory is

$$
S = \int L_{\text{mech}} \, \mathrm{d}t = \int \left(\frac{1}{2}m\dot{\boldsymbol{x}}^2 - V\right) \mathrm{d}t
$$

**Matching.** The optics–mechanics analogy requires the phase of the matter wave to play the role of $\Phi$:

$$
\Phi_{\text{matter}} = \frac{S}{\hbar}
$$

**Verification via de Broglie.** For a free particle with momentum $\boldsymbol{p}$ and energy $E$, the de Broglie wave is $\psi = A \, \mathrm{e}^{\mathrm{i}(\boldsymbol{k} \cdot \boldsymbol{r} - \omega t)}$, so the phase is

$$
\Phi = \boldsymbol{k} \cdot \boldsymbol{r} - \omega t = \frac{\boldsymbol{p} \cdot \boldsymbol{r} - E t}{\hbar}
$$

The classical action for a free particle moving from $\boldsymbol{r}_0$ to $\boldsymbol{r}$ in time $t$ along a straight-line trajectory is $S = \boldsymbol{p} \cdot (\boldsymbol{r} - \boldsymbol{r}_0) - E t$ (up to a constant), confirming $\Phi = S/\hbar$.

**General result.** For any system, the WKB (semiclassical) wavefunction is $\psi = A \, \mathrm{e}^{\mathrm{i} S/\hbar}$, where $S(\boldsymbol{r}, t)$ satisfies the Hamilton-Jacobi equation — the exact classical-to-quantum bridge.
:::

### What is Action?

The action takes different forms across physics:

| **System** | **Action $S$** |
|---|---|
| Light ray | $\int n \, \mathrm{d}s$ |
| Non-relativistic particle | $\int \left(\frac{m\dot{\boldsymbol{x}}^2}{2} - V\right) \mathrm{d}t$ |
| Relativistic particle | $-mc^2 \int \mathrm{d}\tau$ |

In each case, nature chooses the path that makes $S$ stationary: $\delta S = 0$.

### Quantization: From Paths to Probability Amplitudes

Classical mechanics: the observable trajectory satisfies $\delta S = 0$ (Hamilton's principle).

Quantum mechanics: the probability amplitude for a particle to travel from $(\boldsymbol{x}_i, t_i)$ to $(\boldsymbol{x}_f, t_f)$ is obtained by **summing contributions from all possible paths**, each weighted by the phase $\mathrm{e}^{\mathrm{i}S/\hbar}$:

:::{admonition} Path Integral Principle
:class: important

$$
\mathcal{A}(\boldsymbol{x}_i \to \boldsymbol{x}_f) = \int_{\text{all paths}} \mathrm{e}^{\mathrm{i}S[\text{path}]/\hbar} \, \mathcal{D}[\text{path}]
$$

where the integral sums over every conceivable path, weighted by its quantum phase.
:::

**Classical limit** ($S \gg \hbar$): The phase $\mathrm{e}^{\mathrm{i}S/\hbar}$ oscillates rapidly. Neighboring paths have wildly different phases and cancel by destructive interference. Only the stationary-phase path—where $\delta S = 0$—survives. This *is* the classical trajectory.

**Quantum regime** ($S \sim \hbar$): All paths have comparable phases and interfere constructively. Diffraction, tunneling, and superposition emerge naturally.

This is **Huygens' principle for matter**: every point is a source of "matter wavelets" that propagate and interfere. Quantization is the principle of superposing all these wavelets with the correct quantum phase.

:::{admonition} Derivation: Classical Mechanics from Stationary Phase
:class: dropdown information

Start from the path integral:

$$
\mathcal{A}(\boldsymbol{x}_i \to \boldsymbol{x}_f) = \int \mathrm{e}^{\mathrm{i} S[\boldsymbol{x}(t)]/\hbar} \, \mathcal{D}[\boldsymbol{x}]
$$

**Step 1.** Expand $S$ around the classical path $\bar{\boldsymbol{x}}(t)$ (the path satisfying $\delta S = 0$). Write $\boldsymbol{x}(t) = \bar{\boldsymbol{x}}(t) + \boldsymbol{\eta}(t)$ with $\boldsymbol{\eta}(t_i) = \boldsymbol{\eta}(t_f) = 0$:

$$
S[\bar{\boldsymbol{x}} + \boldsymbol{\eta}] = S[\bar{\boldsymbol{x}}] + \underbrace{\delta S}_{ = \, 0} + \frac{1}{2}\delta^2 S + \cdots
$$

The first-order term $\delta S = 0$ vanishes by definition of the classical path (Hamilton's principle). This is the Euler-Lagrange equation:

$$
\frac{\mathrm{d}}{\mathrm{d}t}\frac{\partial L}{\partial \dot{\boldsymbol{x}}} = \frac{\partial L}{\partial \boldsymbol{x}}
$$

**Step 2.** The second-order fluctuation is

$$
\delta^2 S = \int_{t_i}^{t_f} \left(\frac{\partial^2 L}{\partial \dot{x}^2} \dot{\eta}^2 + \frac{\partial^2 L}{\partial x^2} \eta^2\right) \mathrm{d}t
$$

**Step 3.** In the classical limit $S_{\text{cl}} \gg \hbar$, the dominant contribution to the path integral comes from paths near $\bar{\boldsymbol{x}}$. The Gaussian integral over fluctuations $\boldsymbol{\eta}$ gives

$$
\mathcal{A} \approx C \, \mathrm{e}^{\mathrm{i} S_{\text{cl}}/\hbar}
$$

where $C$ is a slowly varying prefactor from the quadratic fluctuations. The probability $|\mathcal{A}|^2$ is dominated by the classical path, and the particle effectively follows the trajectory determined by $\delta S = 0$.

**Correspondence principle.** As $\hbar \to 0$, the Fresnel zone of constructive interference shrinks to zero width around the classical path. Quantum mechanics reduces to classical mechanics, just as wave optics reduces to geometric optics when $\lambda \to 0$.
:::

:::{admonition} Derivation: Hamilton-Jacobi Equation from the WKB Ansatz
:class: dropdown information

Inserting the WKB ansatz $\psi = A\,\mathrm{e}^{\mathrm{i}S/\hbar}$ into the Schrödinger equation shows precisely how classical mechanics emerges from quantum mechanics at $O(\hbar^0)$.

**Step 1: Insert the ansatz.**

The time-dependent Schrödinger equation for a particle in potential $V$ is

$$
\mathrm{i}\hbar\,\partial_t\psi = -\frac{\hbar^2}{2m}\nabla^2\psi + V\psi
$$

With $\psi = A\,\mathrm{e}^{\mathrm{i}S/\hbar}$, compute each term:

$$
\mathrm{i}\hbar\,\partial_t\psi = \left(\mathrm{i}\hbar\,\partial_t A - A\,\partial_t S\right)\mathrm{e}^{\mathrm{i}S/\hbar}
$$

$$
\nabla\psi = \left(\nabla A + \frac{\mathrm{i}}{\hbar}A\,\nabla S\right)\mathrm{e}^{\mathrm{i}S/\hbar}
$$

$$
\nabla^2\psi = \left(\nabla^2 A + \frac{2\mathrm{i}}{\hbar}\nabla A\cdot\nabla S + \frac{\mathrm{i}}{\hbar}A\,\nabla^2 S - \frac{|\nabla S|^2}{\hbar^2}A\right)\mathrm{e}^{\mathrm{i}S/\hbar}
$$

**Step 2: Collect powers of $\hbar$.**

Substituting and dividing by $A\,\mathrm{e}^{\mathrm{i}S/\hbar}$:

$$
\left(\frac{|\nabla S|^2}{2m} + V + \partial_t S\right) + \mathrm{i}\hbar\left(\frac{\partial_t A}{A} + \frac{\nabla A\cdot\nabla S}{mA} + \frac{\nabla^2 S}{2m}\right) - \frac{\hbar^2}{2m}\frac{\nabla^2 A}{A} = 0
$$

**Step 3: $O(\hbar^0)$ — the Hamilton-Jacobi equation.**

The $\hbar^0$ term (real, survives as $\hbar\to 0$) must vanish independently:

$$
\boxed{\partial_t S + \frac{|\nabla S|^2}{2m} + V = 0}
$$

This is the **Hamilton-Jacobi equation** of classical mechanics. With the identification $\boldsymbol{p} = \nabla S$, it becomes $\partial_t S + H(\boldsymbol{r},\boldsymbol{p}) = 0$. Classical trajectories are the integral curves of the velocity field $\boldsymbol{v} = \nabla S/m$ — they are the "rays" of the matter wave.

**Step 4: $O(\hbar^1)$ — probability conservation.**

Setting the $\hbar^1$ imaginary part to zero gives:

$$
\partial_t A + \frac{\nabla A\cdot\nabla S}{m} + \frac{A\,\nabla^2 S}{2m} = 0
$$

Multiplying by $2A$ and using $\rho = A^2$:

$$
\partial_t\rho + \nabla\cdot\!\left(\rho\,\frac{\nabla S}{m}\right) = 0, \qquad \boldsymbol{j} = \rho\,\frac{\nabla S}{m} = \rho\,\boldsymbol{v}
$$

This is the **continuity equation** $\partial_t\rho + \nabla\cdot\boldsymbol{j} = 0$: probability flows along the classical velocity field $\boldsymbol{v} = \boldsymbol{p}/m$.

**Conclusion.** The WKB ansatz splits the Schrödinger equation into two classical structures:
- **Phase** ($\sim\hbar^0$): Hamilton-Jacobi equation — determines the classical trajectory via $\delta S = 0$.
- **Amplitude** ($\sim\hbar^1$): continuity equation — determines how probability spreads along that trajectory.

Classical mechanics is not an approximation tacked onto quantum mechanics; it is the exact leading-order content of the Schrödinger equation.
:::

:::{admonition} Derivation: Gaussian Wave Packet Spreading and Position-Momentum Uncertainty
:class: dropdown information

A Gaussian wave packet provides the cleanest illustration of how the de Broglie relation $p = \hbar k$ combines with Fourier uncertainty to give $\Delta x \Delta p \geq \hbar/2$.

**Step 1: Initial Gaussian.** Consider a wave packet centered at $x = 0$ with momentum $p_0 = \hbar k_0$:

$$
\psi(x, 0) = \left(\frac{1}{2\pi\sigma_0^2}\right)^{1/4} \mathrm{e}^{-x^2/4\sigma_0^2}\, \mathrm{e}^{\mathrm{i}k_0 x}.
$$

The initial position uncertainty is $\Delta x = \sigma_0$.

**Step 2: Momentum space.** Taking the Fourier transform (and using $\int_{-\infty}^\infty \mathrm{e}^{-ax^2+bx}\mathrm{d}x = \sqrt{\pi/a}\,\mathrm{e}^{b^2/4a}$):

$$
\tilde{\psi}(k, 0) = \left(\frac{2\sigma_0^2}{\pi}\right)^{1/4} \mathrm{e}^{-(k-k_0)^2\sigma_0^2}.
$$

This is a Gaussian in $k$ centered at $k_0$ with width $\Delta k = 1/(2\sigma_0)$. Using $p = \hbar k$:

$$
\Delta p = \hbar \Delta k = \frac{\hbar}{2\sigma_0}.
$$

**Step 3: Uncertainty product.** At $t = 0$:

$$
\Delta x \cdot \Delta p = \sigma_0 \cdot \frac{\hbar}{2\sigma_0} = \frac{\hbar}{2}.
$$

The Gaussian wave packet **saturates** the Heisenberg uncertainty bound — it is the minimum-uncertainty state.

**Step 4: Free-particle spreading.** Under free time evolution, each momentum component $k$ acquires phase $\mathrm{e}^{-\mathrm{i}\hbar k^2 t/2m}$, spreading the packet:

$$
\Delta x(t) = \sigma_0 \sqrt{1 + \left(\frac{\hbar t}{2m\sigma_0^2}\right)^2}.
$$

The spread timescale is $t^* = 2m\sigma_0^2/\hbar$. For an electron with $\sigma_0 = 1\,\text{Å}$, $t^* \approx 3 \times 10^{-16}\,\text{s}$ — the wave packet spreads in an attosecond. The position uncertainty grows as $\Delta x \to \hbar t/(2m\sigma_0)$ at late times, consistent with $\Delta p$ remaining constant ($= \hbar/2\sigma_0$): the particle moves with uncertain velocity $\Delta v = \Delta p/m$ and its position grows as $\Delta v \cdot t$.
:::

### Summary

- **Optics–mechanics duality**: Geometric optics and classical mechanics obey identical variational principles. Particles are the "geometric optics" of matter.

- **De Broglie relations**: Matter has wavelength $\lambda = h/p$ and angular frequency $\omega = E/\hbar$.

- **Action = phase**: The quantum wavefunction is $\psi = A \, \mathrm{e}^{\mathrm{i}S/\hbar}$. This unifies classical and quantum descriptions.

- **Path integral quantization**: Sum over all paths with phase $\mathrm{e}^{\mathrm{i}S/\hbar}$. The classical path emerges as the stationary-phase path when $S \gg \hbar$.

- **Classical limit**: Stationary phase selects the classical trajectory. Quantum mechanics automatically reproduces classical mechanics in the appropriate limit.
