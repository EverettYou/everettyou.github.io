## Lecture Notes

### Overview

This section solves the Schrödinger equation for a free particle ($V = 0$) and builds the propagator from plane wave solutions. By superposing plane waves over all momenta and evaluating a Gaussian integral in momentum space, we recover the exact propagator:

$$K_\text{free}(x, t; x_0, 0) = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\frac{\mathrm{i}m}{2\hbar t}(x - x_0)^2\right)$$

The exponent contains the classical action for motion from $x_0$ to $x$ in time $t$. This confirms the central theme: **Phase = Action** is not metaphor—it is exact for the free particle.

### Plane Wave Solution

The free Schrödinger equation (with $V = 0$) is:

$$\mathrm{i}\hbar \frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} \psi$$

A natural ansatz is a **plane wave**:

$$\psi(x, t) = \mathrm{e}^{\mathrm{i}(kx - \omega t)}$$

:::{admonition} Substitute into the Schrödinger equation
:class: dropdown information

$$\mathrm{i}\hbar (-\mathrm{i}\omega) \mathrm{e}^{\mathrm{i}(kx - \omega t)} = -\frac{\hbar^2}{2m} (-k^2) \mathrm{e}^{\mathrm{i}(kx - \omega t)}$$

$$\hbar\omega = \frac{\hbar^2 k^2}{2m}$$
:::

:::{admonition} Plane Wave Dispersion Relation
:class: important

$$\omega(k) = \frac{\hbar k^2}{2m}$$

A plane wave with wave vector $k$ oscillates as $\exp(\mathrm{i}(kx - \omega t))$ and carries energy $E = \hbar\omega = \frac{\hbar^2 k^2}{2m} = \frac{p^2}{2m}$ (momentum eigenstate with $p = \hbar k$).
:::

### Building the Propagator from Plane Waves

The key insight: the propagator is a superposition of all plane wave solutions. A particle can take any momentum, so we sum over all $k$:

$$K_\text{free}(x, t; x_0, 0) = \int \frac{\mathrm{d}k}{(2\pi)^d} \, \mathrm{e}^{\mathrm{i}(k(x - x_0) - \omega(k) t)}$$

$$= \int \frac{\mathrm{d}k}{(2\pi)^d} \exp\left(\mathrm{i}k(x - x_0) - \mathrm{i}\frac{\hbar k^2}{2m} t\right)$$

This is exactly the form anticipated by the path integral: the phase is $S/\hbar$ where $S$ is the action, here summed over all possible momenta.

:::{admonition} Note: Plane Waves as Momentum Eigenstates
:class: note

Each plane wave with momentum $p = \hbar k$ is a momentum eigenstate $\vert p \rangle = (2\pi\hbar)^{-d/2} \mathrm{e}^{\mathrm{i}px/\hbar}$. The propagator is the momentum-space integral of the projector $\vert p \rangle \langle p \vert$ evolved by the time operator $\mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$ with $\hat{H} = \hat{p}^2/(2m)$. This connection to spectral decomposition will be formalized later.
:::

:::{admonition} Derivation: Propagator from Spectral Decomposition
:class: dropdown information

The propagator can be derived rigorously from the operator formalism. By definition,

$$
K_{\text{free}}(x, t; x_0, 0) = \langle x \vert \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar} \vert x_0 \rangle
$$

where $\hat{H} = \hat{p}^2/(2m)$. Insert the momentum resolution of identity $\hat{I} = \int \mathrm{d}p\,\vert p\rangle\langle p\vert$:

$$
K_{\text{free}} = \int \mathrm{d}p\,\langle x \vert \mathrm{e}^{-\mathrm{i}\hat{p}^2 t/(2m\hbar)} \vert p\rangle\langle p \vert x_0\rangle
$$

Since $\vert p\rangle$ is an eigenstate of $\hat{p}$ with eigenvalue $p$, we have $\mathrm{e}^{-\mathrm{i}\hat{p}^2 t/(2m\hbar)}\vert p\rangle = \mathrm{e}^{-\mathrm{i}p^2 t/(2m\hbar)}\vert p\rangle$. The position-space wavefunctions are $\langle x\vert p\rangle = (2\pi\hbar)^{-1/2}\mathrm{e}^{\mathrm{i}px/\hbar}$ and $\langle p\vert x_0\rangle = (2\pi\hbar)^{-1/2}\mathrm{e}^{-\mathrm{i}px_0/\hbar}$. Therefore:

$$
K_{\text{free}} = \int \frac{\mathrm{d}p}{2\pi\hbar}\,\exp\!\left(\frac{\mathrm{i}p(x-x_0)}{\hbar} - \frac{\mathrm{i}p^2 t}{2m\hbar}\right)
$$

Substituting $k = p/\hbar$ (so $\mathrm{d}p = \hbar\,\mathrm{d}k$) recovers exactly

$$
K_{\text{free}} = \int \frac{\mathrm{d}k}{2\pi}\,\exp\!\left(\mathrm{i}k(x-x_0) - \mathrm{i}\frac{\hbar k^2}{2m}t\right)
$$

which is the plane-wave superposition formula. This shows that summing plane waves over all momenta is equivalent to evaluating $\langle x\vert \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}\vert x_0\rangle$ via spectral decomposition.
:::

### Evaluating the Gaussian Integral in Momentum Space

The integral over $k$ is Gaussian; completing the square and shifting variables yields $K_\text{free}$ in closed form.

:::{admonition} The Free-Particle Propagator
:class: important

$$K_\text{free}(x, t; x_0, 0) = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right)$$

This agrees with the one-step free kernel {eq}`eq-K-free-delta-t-one-step` from §3.2.2: set the elapsed time to $\delta t$ (so $t \to \delta t$) and identify endpoints as in time-slicing ($x_0 \leftrightarrow x'$). The Gaussian prefactor and phase then match exactly—**self-consistency** between the short-time path-integral factor and the exact momentum superposition.

:::

:::{admonition} Derivation: Gaussian Evaluation
:class: dropdown information

The exponent is quadratic in $k$:

$$\mathrm{i}k(x - x_0) - \mathrm{i}\frac{\hbar k^2}{2m} t = -\mathrm{i}\frac{\hbar t}{2m}\left(k^2 - \frac{2m}{\hbar t}k(x - x_0)\right)$$

Complete the square in $k$:

$$k^2 - \frac{2m}{\hbar t}k(x - x_0) = \left(k - \frac{m}{\hbar t}(x - x_0)\right)^2 - \frac{m^2}{(\hbar t)^2}(x - x_0)^2$$

The $k$ integral becomes

$$K_\text{free} = \int \frac{\mathrm{d}k}{(2\pi)^d} \exp\left(-\mathrm{i}\frac{\hbar t}{2m}\left[\left(k - \frac{m}{\hbar t}(x - x_0)\right)^2 - \frac{m^2}{(\hbar t)^2}(x - x_0)^2\right]\right)$$

The integral over the shifted quadratic term is a standard Gaussian:

$$\int \mathrm{d}k \exp\left(-\mathrm{i}\frac{\hbar t}{2m} k'^2\right) = \left(\frac{\pi m}{-\mathrm{i}\hbar t/2}\right)^{1/2} = \left(\frac{2\pi m}{\mathrm{i}\hbar t}\right)^{1/2}$$

(where $k' = k - m(x - x_0)/(\hbar t)$ and we use $\sqrt{-\mathrm{i}} = (1-\mathrm{i})/\sqrt{2}$).

The remaining exponent is

$$\exp\left(\mathrm{i}\frac{\hbar t}{2m} \cdot \frac{m^2}{(\hbar t)^2}(x - x_0)^2\right) = \exp\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right)$$

Combining,

$$K_\text{free} = \frac{1}{(2\pi)^d} \left(\frac{2\pi m}{\mathrm{i}\hbar t}\right)^{1/2} \exp\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right) = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right)$$

:::

:::{admonition} Derivation: Composition Property of the Propagator
:class: dropdown information

The propagator satisfies the **composition (Chapman–Kolmogorov) property**: for any intermediate time $t_i < t' < t_f$,

$$
K(x_f, t_f;\, x_i, t_i) = \int_{-\infty}^{\infty} K(x_f, t_f;\, x', t')\, K(x', t';\, x_i, t_i)\, \mathrm{d}x'.
$$

**Step 1: Operator proof.**

The propagator is a matrix element of the time-evolution operator:

$$
K(x_f, t_f; x_i, t_i) = \langle x_f\vert \hat{U}(t_f, t_i)\vert x_i\rangle.
$$

The group property of time evolution gives $\hat{U}(t_f, t_i) = \hat{U}(t_f, t')\hat{U}(t', t_i)$ for any $t' \in (t_i, t_f)$. Inserting the position completeness relation $\int \mathrm{d}x'\,\vert x'\rangle\langle x'\vert = \hat{I}$:

$$
K(x_f, t_f; x_i, t_i) = \langle x_f\vert \hat{U}(t_f, t')\hat{U}(t', t_i)\vert x_i\rangle = \int_{-\infty}^{\infty} \langle x_f\vert \hat{U}(t_f, t')\vert x'\rangle\langle x'\vert \hat{U}(t', t_i)\vert x_i\rangle\,\mathrm{d}x'.
$$

Identifying each factor as a propagator gives the composition formula.

**Step 2: Explicit verification for the free particle.**

We verify that convolving two free propagators over an intermediate time gives the correct total propagator. Define $t_1 = t' - t_i$ and $t_2 = t_f - t'$. The free propagator is (in 1D):

$$
K_\mathrm{free}(x, t; x_0, 0) = \sqrt{\frac{m}{2\pi \mathrm{i}\hbar t}}\exp\!\left(\frac{\mathrm{i}m(x-x_0)^2}{2\hbar t}\right).
$$

We need to evaluate:

$$
I = \int_{-\infty}^{\infty} \sqrt{\frac{m}{2\pi\mathrm{i}\hbar t_2}}\,\mathrm{e}^{\mathrm{i}m(x_f-x')^2/(2\hbar t_2)} \cdot \sqrt{\frac{m}{2\pi\mathrm{i}\hbar t_1}}\,\mathrm{e}^{\mathrm{i}m(x'-x_i)^2/(2\hbar t_1)}\,\mathrm{d}x'.
$$

**Step 3: Complete the square.**

The exponent in the integrand is:

$$
\frac{\mathrm{i}m}{2\hbar}\left[\frac{(x_f-x')^2}{t_2} + \frac{(x'-x_i)^2}{t_1}\right].
$$

Expand and collect terms in $x'$:

$$
= \frac{\mathrm{i}m}{2\hbar}\left[\frac{t_1+t_2}{t_1 t_2}x'^2 - 2\left(\frac{x_f}{t_2}+\frac{x_i}{t_1}\right)x' + \frac{x_f^2}{t_2}+\frac{x_i^2}{t_1}\right].
$$

Let $T = t_1 + t_2$. Complete the square: the $x'$-dependent terms form a Gaussian with center $\bar{x} = (t_1 x_f + t_2 x_i)/T$ and width $\sigma^2 = \mathrm{i}\hbar t_1 t_2/(mT)$.

The Gaussian integral over $x'$ contributes $\sqrt{2\pi\sigma^2} = \sqrt{2\pi\mathrm{i}\hbar t_1 t_2/(mT)}$.

The remaining constant exponent is:

$$
\frac{\mathrm{i}m}{2\hbar}\left[\frac{x_f^2}{t_2}+\frac{x_i^2}{t_1} - \frac{(t_1 x_f+t_2 x_i)^2}{t_1 t_2 T}\right] = \frac{\mathrm{i}m}{2\hbar}\cdot\frac{(x_f-x_i)^2}{T}.
$$

**Step 4: Assemble.**

$$
I = \frac{m}{2\pi\mathrm{i}\hbar\sqrt{t_1 t_2}}\cdot\sqrt{\frac{2\pi\mathrm{i}\hbar t_1 t_2}{mT}}\cdot\exp\!\left(\frac{\mathrm{i}m(x_f-x_i)^2}{2\hbar T}\right) = \sqrt{\frac{m}{2\pi\mathrm{i}\hbar T}}\exp\!\left(\frac{\mathrm{i}m(x_f-x_i)^2}{2\hbar T}\right).
$$

This is exactly $K_\mathrm{free}(x_f, T+t_i; x_i, t_i)$ with total time $T = t_1+t_2$. ✓

**Physical interpretation.** The composition property encodes the quantum version of the **sum over histories**: a particle traveling from $x_i$ to $x_f$ must pass through *some* intermediate point $x'$ at time $t'$, and these intermediate histories are summed (integrated) coherently. It is the mathematical expression of the path integral idea: the amplitude for a journey equals the sum over all ways to complete it.

:::



### Phase = Classical Action

The exponent can be rewritten by recognizing the classical action. A particle traveling from $x_0$ to $x$ in time $t$ with constant velocity has action:

$$S_\text{cl} = \int_0^t \frac{m}{2}\dot{x}^2 \, \mathrm{d}t' = \frac{m}{2} v^2 t = \frac{m}{2t}(x - x_0)^2$$

Thus the exponent in the propagator is precisely $\mathrm{i}S_\text{cl}/\hbar$:

$$K_\text{free} = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\mathrm{i}\frac{S_\text{cl}}{\hbar}\right)$$

:::{admonition} Phase = Action Realized
:class: important

The free-particle propagator embodies the central principle: **the quantum amplitude is the exponential of the classical action divided by $\hbar$**.

The normalization prefactor $(m/2\pi \mathrm{i}\hbar t)^{1/2}$ is the "quantum correction" to the amplitude. The exponent $\mathrm{i}S_\text{cl}/\hbar$ is pure classical action. Together they form the complete quantum propagator.

This is the exact realization of the ansatz from §3.1.3: $\psi \sim \exp(\mathrm{i}S/\hbar)$ is not metaphorical—for the free particle it is mathematically exact.
:::

:::{admonition} Derivation: Initial-Condition Check $K \to \delta(x - x_0)$ as $t \to 0$
:class: dropdown information

A valid propagator must satisfy $K(x, 0; x_0, 0) = \delta(x - x_0)$. We verify this for the free-particle result.

The propagator is a Gaussian in $(x - x_0)$ with width $\sigma^2 = \mathrm{i}\hbar t / m$:

$$
K_{\text{free}}(x, t; x_0, 0) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\!\left(-\frac{(x-x_0)^2}{2\sigma^2}\right), \qquad \sigma^2 = \frac{\mathrm{i}\hbar t}{m}
$$

As $t \to 0^+$, $\vert\sigma\vert \to 0$, so the Gaussian becomes infinitely narrow and tall while maintaining unit normalization:

$$
\int_{-\infty}^{\infty} K_{\text{free}}(x, t; x_0, 0)\,\mathrm{d}x = 1 \quad \text{for all } t > 0
$$

(This follows from the standard Gaussian integral $\int \mathrm{e}^{-\alpha x^2}\mathrm{d}x = \sqrt{\pi/\alpha}$ with $\alpha = m/(2\mathrm{i}\hbar t)$, and the normalization prefactor is chosen to cancel this exactly.)

A normalized function that concentrates to a point is precisely the definition of the delta distribution:

$$
\lim_{t \to 0^+} K_{\text{free}}(x, t; x_0, 0) = \delta(x - x_0)
$$

This confirms the propagator has the correct initial condition. Physically, at $t = 0$ the particle has not moved: the transition amplitude from $x_0$ to $x$ is nonzero only when $x = x_0$.
:::

### Closing the Circle

Recall the structure of unit 3.2:

- **§3.1**: Proposed ψ ∼ exp(iS/ℏ) as the guiding principle
- **§3.2.1**: Formalized the path integral as ∫ 𝒟x exp(iS/ℏ)
- **§3.2.2**: Derived the Schrödinger equation from the path integral (initial value problem)
- **§3.2.3** (now): Solved the Schrödinger equation for the free particle and found K = (normalization) × exp(iS_cl/ℏ)

The propagator closes the loop: the path integral (which sums all classical actions) produces a result whose exponent is exactly the classical action. No approximation. No stationary-phase argument yet. This is exact.

In §3.3 we will use the stationary-phase approximation to extend this insight to general potentials: the dominant classical path contributions to the integral will emerge.

:::{admonition} Derivation: Free-Particle Propagator via Time-Slicing
:class: dropdown information

The spectral method above derives $K_{\text{free}}$ by solving the Schrödinger equation and summing plane waves. Here we verify the same result directly from the path integral, confirming that the two approaches agree.

**Step 1: Set up the time-sliced integral.**
Divide the interval $[0, t]$ into $N$ equal slices of width $\epsilon = t/N$. From §3.2.1, the path integral is

$$
K = \lim_{N\to\infty} \left(\frac{m}{2\pi\mathrm{i}\hbar\epsilon}\right)^{N/2} \int \mathrm{d}x_1\cdots\mathrm{d}x_{N-1}\,\exp\!\left(\frac{\mathrm{i}m}{2\hbar\epsilon}\sum_{k=0}^{N-1}(x_{k+1}-x_k)^2\right)
$$

with $x_0$ and $x_N = x$ fixed.

**Step 2: Evaluate the nested Gaussians iteratively.**
The key identity is: for a Gaussian integral over $x_j$ appearing in two consecutive terms $(x_j - x_{j-1})^2 + (x_{j+1} - x_j)^2$, completing the square in $x_j$ gives

$$
\int_{-\infty}^{\infty} \mathrm{d}x_j\,\exp\!\left(\frac{\mathrm{i}m}{2\hbar\epsilon}\left[(x_j-x_{j-1})^2+(x_{j+1}-x_j)^2\right]\right) = \sqrt{\frac{\mathrm{i}\pi\hbar\epsilon}{m}}\,\exp\!\left(\frac{\mathrm{i}m}{2\hbar(2\epsilon)}(x_{j+1}-x_{j-1})^2\right)
$$

The result is again a Gaussian in the distance $x_{j+1}-x_{j-1}$ but with time interval $2\epsilon$.

**Step 3: Induction.**
After integrating out $x_1$, we have a single Gaussian in $(x_2 - x_0)^2$ with time $2\epsilon$. After integrating out $x_2$, a Gaussian in $(x_3 - x_0)^2$ with time $3\epsilon$. After all $N-1$ integrations, the pattern gives

$$
K = \left(\frac{m}{2\pi\mathrm{i}\hbar\epsilon}\right)^{N/2} \cdot \left(\frac{\mathrm{i}\pi\hbar\epsilon}{m}\right)^{(N-1)/2} \cdot \frac{1}{\sqrt{N}}\,\exp\!\left(\frac{\mathrm{i}m}{2\hbar(N\epsilon)}(x-x_0)^2\right)
$$

**Step 4: Collect prefactors.**
The ratio of prefactors simplifies:

$$
\left(\frac{m}{2\pi\mathrm{i}\hbar\epsilon}\right)^{N/2} \cdot \left(\frac{\mathrm{i}\pi\hbar\epsilon}{m}\right)^{(N-1)/2} = \left(\frac{m}{2\pi\mathrm{i}\hbar\epsilon}\right)^{1/2} \cdot \left(\frac{1}{2}\right)^{(N-1)/2} \cdot \ldots
$$

A cleaner route: track only the final result of the induction. After integrating $k$ intermediate points, the accumulated propagator from $x_0$ to $x_{k+1}$ over time $(k+1)\epsilon$ has the form

$$
\left(\frac{m}{2\pi\mathrm{i}\hbar(k+1)\epsilon}\right)^{1/2}\exp\!\left(\frac{\mathrm{i}m(x_{k+1}-x_0)^2}{2\hbar(k+1)\epsilon}\right)
$$

This is verified by induction: integrate against the one-step kernel using the Gaussian convolution identity from Step 2. Setting $k+1 = N$ and $N\epsilon = t$:

$$
K_{\text{free}}(x,t;x_0,0) = \left(\frac{m}{2\pi\mathrm{i}\hbar t}\right)^{1/2}\exp\!\left(\frac{\mathrm{i}m(x-x_0)^2}{2\hbar t}\right)
$$

This matches the spectral-decomposition result exactly, confirming the consistency between the path integral and Schrödinger equation approaches.
:::

:::{admonition} Discussion: Why Is the Factor i Important?
:class: dropdown tip

The exponent is $+\mathrm{i}S_\text{cl}/\hbar$, not $-\mathrm{i}S_\text{cl}/\hbar$ (as in the path integral formulation itself, which has $\mathcal{S}/\hbar$ with $\mathcal{S}$ defined with a +i sign in the path integral definition). This is crucial:

- A particle with momentum $p = m v$ travels distance $\Delta x = v t = (p/m) t$, so $S_\text{cl} = p \Delta x / 2 = m (\Delta x)^2 / (2t) > 0$.
- With exponent $\mathrm{i}S_\text{cl}/\hbar > 0$, the phase oscillates: $\exp(\mathrm{i}\theta)$ with $\theta = m(\Delta x)^2 / (2\hbar t) \gg 1$ for macroscopic systems. This oscillation suppresses interference from distant paths (destructive interference).
- If the sign were reversed, far paths would constructively interfere—wrong physics.

The +i is the quantum signature of stationary-phase dominance.
:::

### Summary

- **Plane wave**: $\psi = \exp(\mathrm{i}(kx - \omega t))$ with $\omega = \hbar k^2/(2m)$ solves the free Schrödinger equation; each plane wave is a momentum eigenstate.
- **Propagator superposition**: Sum all plane waves over momentum to build the propagator. This is exactly the path integral idea: paths have different momenta.
- **Gaussian evaluation**: Complete the square in momentum space; the integral yields a Gaussian in position with normalization $(m/2\pi \mathrm{i}\hbar t)^{1/2}$.
- **Classical action in exponent**: The exponent equals $\mathrm{i}S_\text{cl}/\hbar$ where $S_\text{cl}$ is the classical action for constant-velocity motion. Phase = Action is exact.
- **Closing the circle**: From the ansatz ψ ∼ exp(iS/ℏ) in §3.1, through path integral quantization in §3.2.1–2, to exact realization in §3.2.3.
