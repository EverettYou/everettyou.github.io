## Lecture Notes

### Overview

The **stationary phase approximation** (SPA) is the mathematical mechanism by which classical mechanics emerges from quantum mechanics. For integrals of the form $I = \int f(x)\, \mathrm{e}^{\mathrm{i}\Phi(x)/\hbar}\,\mathrm{d}x$ with small $\hbar$, the dominant contribution comes from stationary points where $\Phi'(x_0) = 0$. Applied to the path integral, this selects the **classical path** — the trajectory where $\delta S = 0$. This is the correspondence principle made precise.

### The Mathematical Result

Many integrals in quantum mechanics take the form

$$
I = \int \mathrm{e}^{\mathrm{i}\Phi(x)/\hbar} f(x) \, \mathrm{d}x
$$

where $\Phi(x)$ is a phase function, $f(x)$ is a slowly-varying amplitude, and $\hbar$ is small. When $\hbar$ is tiny, the integrand oscillates wildly — contributions cancel everywhere except near points where $\Phi$ varies slowly.

:::{admonition} Stationary Phase Formula
:class: important

For an oscillatory integral with a single stationary point $\Phi'(x_0) = 0$, the leading approximation is

$$
I \approx f(x_0) \sqrt{\frac{2\pi\hbar}{\mathrm{i}\Phi''(x_0)}} \, \mathrm{e}^{\mathrm{i}\Phi(x_0)/\hbar}
$$

Valid when: (1) isolated stationary point with $|\Phi''(x_0)|$ not too small, (2) $f(x)$ slowly varying, (3) $\Phi \gg \hbar$.
:::

:::{admonition} Derivation: Stationary Phase Formula
:class: dropdown information

Near the stationary point $x_0$, expand the phase to second order:

$$
\Phi(x) \approx \Phi(x_0) + \frac{1}{2}\Phi''(x_0)(x - x_0)^2
$$

The integral becomes Gaussian:

$$
I \approx \mathrm{e}^{\mathrm{i}\Phi(x_0)/\hbar} f(x_0) \int \mathrm{e}^{\mathrm{i}\Phi''(x_0)(x-x_0)^2/(2\hbar)} \mathrm{d}x
$$

Using the Fresnel integral $\int \mathrm{e}^{\mathrm{i}ax^2} \mathrm{d}x = \sqrt{\pi/(\mathrm{i}a)}$, the result follows.
:::

:::{admonition} Derivation: Fresnel Gaussian Integral
:class: dropdown information

The key mathematical result underlying the stationary phase approximation is the **Fresnel integral**. Unlike the real Gaussian $\int \mathrm{e}^{-ax^2}\mathrm{d}x = \sqrt{\pi/a}$ ($a>0$), the oscillatory version requires analytic continuation.

**Step 1: Rotate the contour.**

For $a > 0$, consider the integral $I(\epsilon) = \int_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i}(a + \mathrm{i}\epsilon)x^2}\,\mathrm{d}x$ with $\epsilon > 0$. The factor $\mathrm{e}^{-\epsilon x^2}$ ensures convergence, so $I(\epsilon)$ is well-defined.

Substitute $u = \sqrt{a + \mathrm{i}\epsilon}\, x$ (rotating the integration variable into the complex plane):

$$
I(\epsilon) = \frac{1}{\sqrt{a + \mathrm{i}\epsilon}}\int_{\mathcal{C}} \mathrm{e}^{\mathrm{i}u^2}\,\mathrm{d}u
$$

where $\mathcal{C}$ is the ray $\arg(u) = \tfrac{1}{2}\arg(a + \mathrm{i}\epsilon)$, tilted slightly above the real axis.

**Step 2: Deform to the real axis.**

By Cauchy's theorem, the contour $\mathcal{C}$ can be deformed back to the real axis without crossing any singularities (the integrand $\mathrm{e}^{\mathrm{i}u^2}$ is entire). The contributions from the arcs at $|u|\to\infty$ vanish because $\mathrm{e}^{\mathrm{i}u^2}$ decays in the upper half-plane when $\arg(u) \in (0, \pi/2)$. Therefore:

$$
\int_{\mathcal{C}} \mathrm{e}^{\mathrm{i}u^2}\,\mathrm{d}u = \int_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i}u^2}\,\mathrm{d}u
$$

**Step 3: Evaluate via the standard Gaussian.**

The Euler-Poisson trick gives $\int_{-\infty}^\infty \mathrm{e}^{-t^2}\,\mathrm{d}t = \sqrt{\pi}$. Setting $u = \mathrm{e}^{\mathrm{i}\pi/4}t$ (i.e., rotating 45°) transforms $u^2 = \mathrm{i}t^2$, so:

$$
\int_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i}u^2}\,\mathrm{d}u = \mathrm{e}^{\mathrm{i}\pi/4}\int_{-\infty}^{\infty} \mathrm{e}^{-t^2}\,\mathrm{d}t = \sqrt{\pi}\,\mathrm{e}^{\mathrm{i}\pi/4} = \sqrt{\pi\mathrm{i}}
$$

using $\mathrm{e}^{\mathrm{i}\pi/4} = (1+\mathrm{i})/\sqrt{2}$ so $\mathrm{e}^{\mathrm{i}\pi/2} = \mathrm{i}$ and $\sqrt{\pi}\cdot\mathrm{e}^{\mathrm{i}\pi/4} = \sqrt{\pi}\cdot\sqrt{\mathrm{i}} = \sqrt{\pi\mathrm{i}}$.

**Step 4: Take $\epsilon \to 0^+$.**

$$
I(\epsilon) = \frac{\sqrt{\pi\mathrm{i}}}{\sqrt{a + \mathrm{i}\epsilon}} \xrightarrow{\epsilon\to 0^+} \frac{\sqrt{\pi\mathrm{i}}}{\sqrt{a}} = \sqrt{\frac{\pi\mathrm{i}}{a}} = \sqrt{\frac{\pi}{\mathrm{i}a}}\cdot\mathrm{i} = \sqrt{\frac{\pi}{-\mathrm{i}a}\cdot(-1)} 
$$

More directly: $\sqrt{\pi\mathrm{i}/a} = \sqrt{\pi/(a/\mathrm{i})} = \sqrt{\pi\mathrm{i}/a}$. Using $1/\mathrm{i} = -\mathrm{i}$:

$$
\boxed{\int_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i}ax^2}\,\mathrm{d}x = \sqrt{\frac{\pi}{\mathrm{i}a}}, \quad a > 0}
$$

The factor $1/\sqrt{\mathrm{i}a}$ combines a magnitude $1/\sqrt{a}$ (the "width" of the oscillation region) and a phase $\mathrm{e}^{-\mathrm{i}\pi/4}$ (the **Maslov phase** shift from the 45° rotation).

**Extension to $a < 0$:** Replace $a \to |a|\mathrm{e}^{\mathrm{i}\pi}$ and rotate the contour by $-\pi/4$ instead, giving $\sqrt{\pi/|\mathrm{i}a|} = \sqrt{\pi/(\mathrm{i}a)}$ with $a < 0$ (now $1/(\mathrm{i}a) > 0$ since $a<0$). The formula $\sqrt{\pi/(\mathrm{i}a)}$ holds for all real $a \neq 0$, with the square root branch chosen to have positive real part.
:::

:::{admonition} Derivation: Detailed Stationary Phase with Higher-Order Corrections
:class: dropdown information

We give a more careful derivation showing how the error term arises.

**Setup.** Consider $I = \int f(x)\,\mathrm{e}^{\mathrm{i}\Phi(x)/\hbar}\,\mathrm{d}x$ with $\Phi'(x_0) = 0$ and $\Phi''(x_0) \neq 0$.

**Step 1: Localize.** Substitute $x = x_0 + u$ and expand:

$$
\Phi(x_0 + u) = \Phi(x_0) + \tfrac{1}{2}\Phi''(x_0)\,u^2 + \tfrac{1}{6}\Phi'''(x_0)\,u^3 + \tfrac{1}{24}\Phi^{(4)}(x_0)\,u^4 + \cdots
$$

$$
f(x_0 + u) = f(x_0) + f'(x_0)\,u + \tfrac{1}{2}f''(x_0)\,u^2 + \cdots
$$

**Step 2: Gaussian integral.** The leading term is

$$
I_0 = f(x_0)\,\mathrm{e}^{\mathrm{i}\Phi(x_0)/\hbar} \int \mathrm{e}^{\mathrm{i}\Phi''(x_0)\,u^2/(2\hbar)}\,\mathrm{d}u = f(x_0)\,\mathrm{e}^{\mathrm{i}\Phi(x_0)/\hbar}\sqrt{\frac{2\pi\hbar}{\mathrm{i}\Phi''(x_0)}}
$$

where we used the Fresnel integral $\int_{-\infty}^{\infty}\mathrm{e}^{\mathrm{i}au^2}\,\mathrm{d}u = \sqrt{\pi/(\mathrm{i}a)}$ with $a = \Phi''(x_0)/(2\hbar)$.

**Step 3: Corrections.** The next corrections come from:

- The $u^3$ term in $\Phi$: contributes at $O(\hbar^{1/2})$ but vanishes by oddness of $u^3 \cdot \mathrm{e}^{\mathrm{i}au^2}$.
- The $u^4$ term in $\Phi$ crossed with the leading $f(x_0)$: contributes at $O(\hbar)$.
- The $f'(x_0)\,u$ term: vanishes by oddness.
- The $f''(x_0)\,u^2/2$ term: contributes at $O(\hbar)$.

The first nonvanishing correction is $O(\hbar)$ relative to $I_0$:

$$
I = I_0\left[1 + O\!\left(\frac{\hbar\,\Phi^{(4)}}{(\Phi'')^2}\right) + O\!\left(\frac{\hbar\,f''}{f\,\Phi''}\right)\right]
$$

The SPA is accurate when $\hbar$ is small compared to the scales set by $\Phi''$, $\Phi^{(4)}$, and the variation of $f$.

**Step 4: Square root branch.** The square root $\sqrt{1/(\mathrm{i}\Phi'')}$ requires a branch choice. For $\Phi'' > 0$, $\sqrt{1/\mathrm{i}} = \mathrm{e}^{-\mathrm{i}\pi/4}$; for $\Phi'' < 0$, $\sqrt{1/(-\mathrm{i})} = \mathrm{e}^{+\mathrm{i}\pi/4}$. This phase is the **Maslov index** contribution, important in semiclassical quantization (§3.3.3).
:::

The width of the region that contributes is $\Delta x \sim \sqrt{\hbar/|\Phi''|}$. As $\hbar \to 0$, this region shrinks to zero — only the stationary point survives.

### Why the Classical Path Dominates

In the path integral

$$
K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int \mathrm{e}^{\mathrm{i}S[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]
$$

the phase function is the classical action $S[\boldsymbol{x}]$. The stationary point is the **classical path** $\boldsymbol{x}_\mathrm{cl}(t)$, defined by $\delta S/\delta \boldsymbol{x} = 0$ — the Euler-Lagrange (Newton's) equations.

**Constructive interference at $\boldsymbol{x}_\mathrm{cl}$:** Nearby paths have actions differing only at second order, so their phases nearly agree — they add coherently.

**Destructive interference away from $\boldsymbol{x}_\mathrm{cl}$:** Paths far from the classical trajectory have first-order action differences $\Delta S \gg \hbar$, producing rapidly oscillating phases that cancel.

:::{admonition} Discussion: Classical Emergence
:class: dropdown tip

The stationary phase approximation says that as $\hbar \to 0$, only classical paths survive. This raises several questions:

- In the limit $\hbar \to 0$, does stationary phase become *exact* or just increasingly accurate? What controls the error?
- For a potential with multiple classical paths (e.g., reflection and transmission), do they interfere constructively or destructively?
- Tunneling has no classical path, yet it occurs. How does the path integral handle contributions from classically forbidden trajectories?
:::

### Quantum Fluctuations: The Second Variation

Expanding the action to second order around the classical path:

$$
S[\boldsymbol{x}_\mathrm{cl} + \delta\boldsymbol{x}] \approx S_\mathrm{cl} + \frac{1}{2}\int \mathrm{d}t \, \delta\boldsymbol{x} \cdot M \cdot \delta\boldsymbol{x}
$$

where $M = -m\frac{\mathrm{d}^2}{\mathrm{d}t^2} - V''(\boldsymbol{x}_\mathrm{cl})$ is the second-variation operator. The Gaussian integral over fluctuations gives the **semiclassical propagator**:

:::{admonition} Semiclassical Propagator
:class: important

$$
K \approx \mathrm{e}^{\mathrm{i}S_\mathrm{cl}/\hbar} \cdot \frac{1}{\sqrt{\det M}}
$$

The phase comes from the classical action; the prefactor encodes quantum fluctuations.
:::

:::{admonition} Derivation: Fluctuation Determinant from Gaussian Functional Integral
:class: dropdown information

We derive the prefactor $1/\sqrt{\det M}$ in the semiclassical propagator.

**Step 1: Discretize.** Write $\delta x(t) = \sum_n c_n \phi_n(t)$ where $\{\phi_n\}$ are orthonormal eigenfunctions of the second-variation operator $M$:

$$
M\,\phi_n = \lambda_n\,\phi_n, \qquad \int \phi_n(t)\,\phi_m(t)\,\mathrm{d}t = \delta_{nm}
$$

with boundary conditions $\phi_n(t_i) = \phi_n(t_f) = 0$ (fluctuations vanish at the endpoints).

**Step 2: Diagonalize.** In these coordinates, the second variation becomes

$$
\frac{1}{2}\int \delta x \cdot M \cdot \delta x\,\mathrm{d}t = \frac{1}{2}\sum_n \lambda_n\,c_n^2
$$

The functional integral factorizes into a product of ordinary Gaussian integrals:

$$
\int \mathcal{D}[\delta x]\,\exp\!\left(\frac{\mathrm{i}}{2\hbar}\sum_n \lambda_n\,c_n^2\right) = \prod_n \int \mathrm{d}c_n\,\exp\!\left(\frac{\mathrm{i}\lambda_n}{2\hbar}\,c_n^2\right)
$$

**Step 3: Evaluate.** Each Gaussian integral gives $\sqrt{2\pi\hbar/(\mathrm{i}\lambda_n)}$, so the product is

$$
\prod_n \sqrt{\frac{2\pi\hbar}{\mathrm{i}\lambda_n}} = \left(\frac{2\pi\hbar}{\mathrm{i}}\right)^{N/2}\frac{1}{\sqrt{\prod_n \lambda_n}} = \frac{\text{const}}{\sqrt{\det M}}
$$

where $\det M = \prod_n \lambda_n$ is the functional determinant (regularized by the ratio with a reference operator).

**Step 4: Combine.** Including the classical phase:

$$
K \approx \left(\frac{1}{2\pi\mathrm{i}\hbar}\right)^{1/2}\frac{1}{\sqrt{\det M / \det M_{\text{free}}}}\,\mathrm{e}^{\mathrm{i}S_{\text{cl}}/\hbar}
$$

The ratio $\det M / \det M_{\text{free}}$ ensures proper normalization — the free-particle case ($V = 0$) reproduces the known result $K_{\text{free}} = (m/2\pi\mathrm{i}\hbar t)^{1/2}\,\mathrm{e}^{\mathrm{i}S_{\text{cl}}/\hbar}$.

This procedure — expand to second order, diagonalize, evaluate Gaussian integrals — is the prototype for all semiclassical and perturbative calculations in quantum field theory.
:::

:::{admonition} Example: Electron vs Baseball
:class: dropdown example

**Problem.** Compare $S/\hbar$ for an electron moving 1 cm in 1 s versus a baseball (0.1 kg) moving 1 m in 1 s.

**Solution.** The classical action is $S \sim md^2/t$.

- Electron: $S \sim (10^{-30})(10^{-2})^2/1 = 10^{-34}$ J·s $\approx \hbar$. Quantum regime — all paths matter.
- Baseball: $S \sim (0.1)(1)^2/1 = 0.1$ J·s $= 10^{33}\hbar$. Classical regime — only the classical path survives.
:::

### The Correspondence Principle

Stationary phase quantifies the **correspondence principle**: quantum mechanics reduces to classical mechanics when $S \gg \hbar$.

| Regime | $S/\hbar$ | Behavior |
|--------|-----------|----------|
| Classical | $\gg 1$ | Single path dominates; definite trajectory |
| Semiclassical | $\sim 1$ | Classical path + quantum corrections |
| Quantum | $\lesssim 1$ | All paths contribute; wave-like behavior |

:::{admonition} Derivation: Multiple Stationary Points and Semiclassical Interference
:class: dropdown information

When a physical amplitude receives contributions from **two or more classical paths**, the semiclassical approximation is a superposition. The resulting interference encodes the quantum mechanical principle that probability amplitudes — not probabilities — add.

**Setup.** Suppose the path integral $K(x_f, x_i; t) = \int \mathrm{e}^{\mathrm{i}S[x]/\hbar}\mathcal{D}[x]$ has two stationary-phase paths $x_\alpha(t)$ ($\alpha=1,2$), each satisfying $\delta S = 0$. The semiclassical propagator is

$$
K \approx \sum_{\alpha=1}^2 C_\alpha\, \mathrm{e}^{\mathrm{i}S_{cl}^{(\alpha)}/\hbar - \mathrm{i}\pi\nu_\alpha/2},
$$

where $S_{cl}^{(\alpha)}$ is the classical action along path $\alpha$, $C_\alpha = \sqrt{|{-}\partial^2 S_{cl}^{(\alpha)}/\partial x_f\partial x_i|/(2\pi\mathrm{i}\hbar)}$ is the fluctuation amplitude (Van Vleck factor), and $\nu_\alpha$ is the Maslov index (number of conjugate points, contributing phase $-\pi/2$ each).

**Step 1: Detection probability.**

The probability density at $x_f$ is $|K|^2 \approx |C_1|^2 + |C_2|^2 + 2\,\mathrm{Re}\bigl(C_1 C_2^*\, \mathrm{e}^{\mathrm{i}\Delta S/\hbar}\bigr)$, where $\Delta S = S_{cl}^{(1)} - S_{cl}^{(2)} - \tfrac{\pi\hbar}{2}(\nu_1 - \nu_2)$. The **cross term** is the interference; it oscillates between constructive and destructive as the phase difference $\Delta S/\hbar$ varies.

**Step 2: Double-slit as the canonical example.**

Two slits at positions $\pm d/2$ (separation $d$), screen at distance $L$. A particle of momentum $p = mv = h/\lambda_\mathrm{dB}$ travels from source to slit $\alpha$, then to the screen at position $y$.

The classical action for each path is (approximately, for $y \ll L$):

$$
S_{cl}^{(\alpha)} = p \cdot r_\alpha, \quad r_\alpha = \sqrt{L^2 + (y \mp d/2)^2} \approx L + \frac{(y \mp d/2)^2}{2L}.
$$

The action difference is

$$
\Delta S = S_{cl}^{(1)} - S_{cl}^{(2)} \approx p\cdot\frac{-2y\cdot d/2}{L} \cdot \frac{1}{1} = \frac{-pyd}{L} = -\frac{2\pi d\sin\theta}{\lambda_\mathrm{dB}},
$$

where $\sin\theta \approx y/L$ is the diffraction angle.

**Step 3: Interference pattern.**

Both paths have equal Maslov index and equal prefactors $|C_1| = |C_2| \equiv C$ (by symmetry). The intensity is

$$
I(y) \propto |C\,\mathrm{e}^{\mathrm{i}S_1/\hbar} + C\,\mathrm{e}^{\mathrm{i}S_2/\hbar}|^2 = 2C^2\bigl(1 + \cos(\Delta S/\hbar)\bigr) = 4C^2\cos^2\!\left(\frac{\pi d\sin\theta}{\lambda_\mathrm{dB}}\right).
$$

This is the standard double-slit fringe formula. **Constructive interference** occurs when $d\sin\theta = n\lambda_\mathrm{dB}$ (integer $n$); **destructive** when $d\sin\theta = (n+\tfrac{1}{2})\lambda_\mathrm{dB}$.

**Step 4: Why classical mechanics does not interfere.**

In classical mechanics, probabilities — not amplitudes — add: $P = P_1 + P_2 = 2C^2$, independent of $y$. The interference term appears only because we sum complex amplitudes $\mathrm{e}^{\mathrm{i}S/\hbar}$ before squaring. The fringe visibility goes to zero as $\hbar \to 0$ faster than the fringe spacing shrinks — in the classical limit, fringes become unresolvably dense and wash out, leaving the uniform $2C^2$ classical answer. This is how the semiclassical limit recovers classical probability from quantum interference.
:::

### Summary

- **Stationary phase**: oscillatory integrals are dominated by points where $\Phi'(x_0) = 0$; the contributing region has width $\sim \sqrt{\hbar}$
- **Classical path selection**: the path integral selects $\delta S = 0$ through constructive interference; all other paths cancel
- **Semiclassical propagator**: $K \approx \mathrm{e}^{\mathrm{i}S_\mathrm{cl}/\hbar}/\sqrt{\det \hat{D}}$ includes quantum corrections from the second variation
- **Correspondence principle**: classical mechanics emerges when $S \gg \hbar$; the ratio $S/\hbar$ controls the quantum-to-classical transition
