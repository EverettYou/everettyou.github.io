## Lecture Notes

### Overview

Quantum mechanics places fundamental limits on the joint knowledge we can have of two observables. These limits arise not from measurement error, but from the non-commutativity of the corresponding operators. This section defines variance and expectation value, introduces the commutator, proves that commuting observables share an eigenbasis, and derives the Heisenberg uncertainty principle.

### Expectation Value and Variance

For a measurement outcome that is random, we characterize the distribution by its expectation value and variance. If observable $\hat{O}$ has eigenstates $\vert O=m \rangle$ with eigenvalues $m$, and state $\vert\psi\rangle = \sum_m \psi_m \vert O=m \rangle$, then:

$$

\langle\hat{O}\rangle_\psi = \sum_m m \, P(m \vert \psi, \hat{O})

$$

where $P(m \vert \psi, \hat{O}) = |\psi_m|^2$ is the probability of outcome $m$.

:::{admonition} Expectation Value and Variance
:class: important

For a Hermitian operator $\hat{O}$ and state $\vert\psi\rangle$:

$$

\langle\hat{O}\rangle_\psi = \langle\psi\vert\hat{O}\vert\psi\rangle

$$

$$

\mathrm{Var}(\hat{O}) = \langle\hat{O}^2\rangle - \langle\hat{O}\rangle^2

$$

$$

\Delta\hat{O} = \sqrt{\mathrm{Var}(\hat{O})}

$$

The uncertainty $\Delta\hat{O}$ is the standard deviation of outcomes.
:::

:::{admonition} Derivation: $\langle\hat{O}\rangle_\psi = \langle\psi\vert\hat{O}\vert\psi\rangle$
:class: dropdown information

Start from the spectral decomposition $\hat{O} = \sum_m m \, \hat{P}_{O=m} = \sum_m m \vert O{=}m\rangle\langle O{=}m\vert$ and the Born rule $P(m\vert\psi) = |\psi_m|^2$:

$$

\langle\hat{O}\rangle_\psi = \sum_m m \, P(m\vert\psi) = \sum_m m \, |\psi_m|^2

$$

Expand $\vert\psi\rangle = \sum_{m'} \psi_{m'} \vert O{=}m'\rangle$ and compute $\langle\psi\vert\hat{O}\vert\psi\rangle$ directly:

$$

\langle\psi\vert\hat{O}\vert\psi\rangle = \sum_{m',m''} \psi_{m'}^* \psi_{m''} \langle O{=}m'\vert \hat{O} \vert O{=}m''\rangle = \sum_{m',m''} \psi_{m'}^* \psi_{m''} \, m'' \, \delta_{m'm''} = \sum_m m \, |\psi_m|^2

$$

which is the same expression. $\square$
:::

:::{admonition} Derivation: Variance Formula
:class: dropdown information

We show that $\mathrm{Var}(\hat{O}) = \langle(\hat{O} - \langle\hat{O}\rangle)^2\rangle = \langle\hat{O}^2\rangle - \langle\hat{O}\rangle^2$.

Define the centered operator $\Delta\hat{O} = \hat{O} - \langle\hat{O}\rangle\hat{I}$, where $\langle\hat{O}\rangle$ is a scalar. Then:

$$

\langle(\Delta\hat{O})^2\rangle = \langle(\hat{O} - \langle\hat{O}\rangle)^2\rangle = \langle\hat{O}^2 - 2\langle\hat{O}\rangle\hat{O} + \langle\hat{O}\rangle^2\rangle

$$

Using linearity of the expectation value:

$$

= \langle\hat{O}^2\rangle - 2\langle\hat{O}\rangle\langle\hat{O}\rangle + \langle\hat{O}\rangle^2 = \langle\hat{O}^2\rangle - \langle\hat{O}\rangle^2

$$

This is non-negative because $\langle(\Delta\hat{O})^2\rangle = \langle\psi\vert(\Delta\hat{O})^\dagger\Delta\hat{O}\vert\psi\rangle = \|\Delta\hat{O}\vert\psi\rangle\|^2 \geq 0$, using the fact that $\Delta\hat{O}$ is Hermitian. Equality holds if and only if $\vert\psi\rangle$ is an eigenstate of $\hat{O}$. $\square$
:::

### The Commutator

When two operators $\hat{A}$ and $\hat{B}$ applied in different orders yield different results, we measure this non-commutativity by the commutator:

:::{admonition} Commutator
:class: important

$$

[\hat{A}, \hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}

$$

If $[\hat{A}, \hat{B}] = 0$, the operators **commute**.
:::

**Physical meaning:** The commutator measures whether operator application order matters. In classical mechanics, all observables commute. In quantum mechanics, non-commutativity reflects a deep constraint: we cannot simultaneously sharpen our knowledge of both observables.

:::{admonition} Analogy: Commute = Independent
:class: note

Think of getting dressed. Let 
- $\hat{A}$ = put on socks, 
- $\hat{B}$ = put on shoes, 
- $\hat{C}$ = put on a hat. 

Then $\hat{A}\hat{B} \neq \hat{B}\hat{A}$ (socks then shoes is different from shoes then socks), but $\hat{A}\hat{C} = \hat{C}\hat{A}$ (hat doesn't care about the socks-shoes order). The hat and socks/shoes *commute* because they are **independent**.

:::

**Key properties of the commutator:**
- Linearity: $[\hat{A}, \hat{B} + \hat{C}] = [\hat{A}, \hat{B}] + [\hat{A}, \hat{C}]$
- Antisymmetry: $[\hat{A}, \hat{B}] = -[\hat{B}, \hat{A}]$
- Jacobi identity: $[\hat{A}, [\hat{B}, \hat{C}]] + [\hat{B}, [\hat{C}, \hat{A}]] + [\hat{C}, [\hat{A}, \hat{B}]] = 0$
- Product rule: $[\hat{A}, \hat{B}\hat{C}] = [\hat{A}, \hat{B}]\hat{C} + \hat{B}[\hat{A}, \hat{C}]$

### Commuting Operators Share Eigenbasis

A central theorem relates commutativity to shared eigenstates:

:::{admonition} Theorem: Commuting Operators
:class: important

Two Hermitian operators $\hat{A}$ and $\hat{B}$ commute, i.e. $[\hat{A}, \hat{B}] = 0$, if and only if they share a complete set of simultaneous eigenstates.

**Equivalently**: If $[\hat{A}, \hat{B}] = 0$, we can measure both $\hat{A}$ and $\hat{B}$ simultaneously with arbitrary precision on a single copy of the state.
:::

:::{admonition} Proof
:class: dropdown information

Suppose $[\hat{A}, \hat{B}] = 0$ and $\vert u\rangle$ is an eigenstate of $\hat{A}$ with eigenvalue $a$:

$$

\hat{A}\vert u\rangle = a\vert u\rangle

$$

Then:

$$

\hat{A}(\hat{B}\vert u\rangle) = \hat{B}\hat{A}\vert u\rangle = \hat{B}(a\vert u\rangle) = a(\hat{B}\vert u\rangle)

$$

So $\hat{B}\vert u\rangle$ is also an eigenstate of $\hat{A}$ with the same eigenvalue $a$. If the eigenvalue $a$ is non-degenerate, then $\hat{B}\vert u\rangle = b\vert u\rangle$ for some $b$, making $\vert u\rangle$ an eigenstate of both. For degenerate eigenvalues, $\hat{B}$ acts within the degenerate subspace, and we can diagonalize it there. Iterating this construction yields a basis diagonalizing both $\hat{A}$ and $\hat{B}$ simultaneously.

Conversely, if $\vert u_j\rangle$ are simultaneous eigenstates with $\hat{A}\vert u_j\rangle = a_j\vert u_j\rangle$ and $\hat{B}\vert u_j\rangle = b_j\vert u_j\rangle$, then on this basis $\hat{A}$ and $\hat{B}$ are diagonal matrices, which commute.
:::

### Heisenberg Uncertainty Principle

The physical consequence of non-commutativity is the uncertainty principle:

:::{admonition} Robertson Uncertainty Relation
:class: important

For any state $\vert\psi\rangle$ and any two observables $\hat{A}$, $\hat{B}$:

$$

\Delta\hat{A} \cdot \Delta\hat{B} \geq \frac{1}{2}|\langle[\hat{A}, \hat{B}]\rangle|

$$

where $\Delta\hat{O} = \sqrt{\langle\hat{O}^2\rangle - \langle\hat{O}\rangle^2}$ and $\langle[\hat{A}, \hat{B}]\rangle = \langle\psi\vert[\hat{A}, \hat{B}]\vert\psi\rangle$.
:::

**Interpretation:** When two observables do not commute, the product of their uncertainties has a positive lower bound set by their commutator. The more strongly they fail to commute, the larger the lower bound on the product of uncertainties.

:::{admonition} Important: Repeated Identical Measurements
:class: note

The uncertainty principle describes repeated measurements on **identical copies** of the same prepared state, not sequential measurement of a single copy. If you measure $\hat{A}$ first and then $\hat{B}$ on the same particle, you collapse the state and $\hat{B}$'s outcome depends on $\hat{A}$'s result—that is a different story (back-action). Here we imagine preparing the state fresh many times and analyzing the distribution across those trials.
:::

:::{admonition} Proof: Robertson Uncertainty Relation
:class: dropdown information

Define $\Delta\hat{A} = \hat{A} - \langle\hat{A}\rangle$ and similarly for $\hat{B}$. These are centered operators with zero expectation value.

By Cauchy-Schwarz inequality,

$$

\|\Delta\hat{A}\vert\psi\rangle\|^2 \cdot \|\Delta\hat{B}\vert\psi\rangle\|^2 \geq |\langle\psi\vert(\Delta\hat{A})^\dagger\Delta\hat{B}\vert\psi\rangle|^2

$$

Since $\hat{A}$ and $\hat{B}$ are Hermitian, so are $\Delta\hat{A}$ and $\Delta\hat{B}$, and:

$$

\langle\psi\vert(\Delta\hat{A})^\dagger\Delta\hat{B}\vert\psi\rangle = \langle\psi\vert\Delta\hat{A}\Delta\hat{B}\vert\psi\rangle

$$

Decompose into symmetric and antisymmetric parts:

$$

\Delta\hat{A}\Delta\hat{B} = \frac{1}{2}\{\Delta\hat{A}, \Delta\hat{B}\} + \frac{1}{2}[\Delta\hat{A}, \Delta\hat{B}]

$$

The anticommutator $\{\Delta\hat{A}, \Delta\hat{B}\}$ is Hermitian (product of two Hermitian operators, symmetrized), so $\langle\psi\vert\tfrac{1}{2}\{\Delta\hat{A}, \Delta\hat{B}\}\vert\psi\rangle$ is **real**. The commutator $[\Delta\hat{A}, \Delta\hat{B}]$ is anti-Hermitian ($[\hat{A},\hat{B}]^\dagger = -[\hat{A},\hat{B}]$ for Hermitian $\hat{A},\hat{B}$), so $\langle\psi\vert\tfrac{1}{2}[\Delta\hat{A}, \Delta\hat{B}]\vert\psi\rangle$ is **purely imaginary**. Writing $z = \langle\psi\vert\Delta\hat{A}\Delta\hat{B}\vert\psi\rangle$:

$$

|z|^2 = (\operatorname{Re} z)^2 + (\operatorname{Im} z)^2 \geq (\operatorname{Im} z)^2 = \left|\frac{1}{2}\langle[\Delta\hat{A}, \Delta\hat{B}]\rangle\right|^2

$$

Finally, note that $[\Delta\hat{A}, \Delta\hat{B}] = [\hat{A} - \langle\hat{A}\rangle, \hat{B} - \langle\hat{B}\rangle] = [\hat{A}, \hat{B}]$ (scalar shifts cancel in commutators). Therefore:

$$

|\langle\psi\vert\Delta\hat{A}\Delta\hat{B}\vert\psi\rangle|^2 \geq \frac{1}{4}|\langle[\hat{A}, \hat{B}]\rangle|^2

$$

Therefore:

$$

(\Delta\hat{A})^2(\Delta\hat{B})^2 \geq \frac{1}{4}|\langle[\hat{A}, \hat{B}]\rangle|^2

$$

Taking square roots gives the Robertson relation.
:::

:::{admonition} Example: Uncertainty Relation for Pauli Operators
:class: dropdown example

For a qubit with Pauli operators:

$$
[\hat{\sigma}^x, \hat{\sigma}^z] = -2\mathrm{i}\hat{\sigma}^y
$$

In state $\vert\psi\rangle = \vert 0\rangle$ (eigenstate of $\hat{Z}$):
- $\langle\hat{Z}\rangle = 1$, so $\Delta\hat{Z} = 0$
- $\langle\hat{X}\rangle = 0$, and $\langle\hat{X}^{2}\rangle = 1$, so $(\Delta\hat{X})^2 = 1$
- The bound from Robertson: $\Delta\hat{X} \cdot \Delta\hat{Z} \geq \frac{1}{2}\vert\langle[\hat{X}, \hat{Z}]\rangle\vert = \frac{1}{2}\vert\langle -2\mathrm{i}\hat{Y}\rangle\vert = 0$

The bound is **not violated** because $\Delta\hat{\sigma}^z = 0$ saturates the lower bound.
:::


:::{admonition} Derivation: Minimum Uncertainty States
:class: dropdown information

The Robertson bound $\Delta\hat{A}\cdot\Delta\hat{B} \geq \tfrac{1}{2}\vert\langle[\hat{A},\hat{B}]\rangle\vert$ is generally strict. We find the precise condition under which equality holds — the **minimum uncertainty states**.

**Step 1: Two inequalities in the Robertson proof.**

Set $\vert f\rangle = (\hat{A} - \langle\hat{A}\rangle)\vert\psi\rangle$ and $\vert g\rangle = (\hat{B} - \langle\hat{B}\rangle)\vert\psi\rangle$. The proof uses:

1. Cauchy-Schwarz: $\langle f\vert f\rangle\langle g\vert g\rangle \geq \vert\langle f\vert g\rangle\vert^2$
2. Dropping the real part: $\vert\langle f\vert g\rangle\vert^2 \geq \bigl(\mathrm{Im}\langle f\vert g\rangle\bigr)^2 = \tfrac{1}{4}\vert\langle[\hat{A},\hat{B}]\rangle\vert^2$

Both inequalities must be equalities simultaneously for the bound to be saturated.

**Step 2: Cauchy-Schwarz equality condition.**

Cauchy-Schwarz is an equality if and only if the two vectors are proportional:

$$
\vert f\rangle = \mu\,\vert g\rangle \quad \text{for some complex scalar } \mu
$$

i.e., $(\hat{A} - \langle\hat{A}\rangle)\vert\psi\rangle = \mu\,(\hat{B} - \langle\hat{B}\rangle)\vert\psi\rangle$.

**Step 3: Condition from ignoring the real part.**

Since $\langle f\vert g\rangle = \mu\langle g\vert g\rangle = \mu(\Delta\hat{B})^2$, dropping $(\mathrm{Re}\langle f\vert g\rangle)^2$ requires $\mathrm{Re}\langle f\vert g\rangle = 0$, which forces $\mu$ to be **purely imaginary**: $\mu = \mathrm{i}\lambda$ with $\lambda \in \mathbb{R}$.

**Step 4: The minimum-uncertainty eigenvalue equation.**

Combining both conditions: $\vert\psi\rangle$ saturates the Robertson bound if and only if

$$
\bigl(\hat{A} - \langle\hat{A}\rangle + \mathrm{i}\lambda(\hat{B} - \langle\hat{B}\rangle)\bigr)\vert\psi\rangle = 0
$$

for some real $\lambda$. The ratio of uncertainties satisfies $\lambda = \Delta\hat{A}/\Delta\hat{B}$.

**Step 5: Gaussian wavepackets for position and momentum.**

Apply this to $\hat{A} = \hat{x}$ and $\hat{B} = \hat{p} = -\mathrm{i}\hbar\partial_x$, for which $[\hat{x}, \hat{p}] = \mathrm{i}\hbar$. The eigenvalue equation becomes:

$$
(x - x_0)\psi + \mathrm{i}\lambda\!\left(-\mathrm{i}\hbar\partial_x - p_0\right)\psi = 0 \quad \Longrightarrow \quad \partial_x\psi = \left(\frac{\mathrm{i}p_0}{\hbar} - \frac{x - x_0}{\lambda\hbar}\right)\psi
$$

Integrating: $\ln\psi = \mathrm{i}p_0 x/\hbar - (x - x_0)^2/(2\lambda\hbar) + \mathrm{const}$, so with $\lambda = 2\sigma^2/\hbar > 0$,

$$
\psi(x) = \left(\frac{1}{2\pi\sigma^2}\right)^{1/4}\exp\!\left(-\frac{(x-x_0)^2}{4\sigma^2} + \frac{\mathrm{i}p_0 x}{\hbar}\right)
$$

This is a **Gaussian wavepacket** centered at $x_0$ with momentum $p_0$. Direct computation gives $\Delta x = \sigma$ and $\Delta p = \hbar/(2\sigma)$, so $\Delta x \cdot \Delta p = \hbar/2$ — exactly the Heisenberg limit.

**Physical interpretation.** Minimum uncertainty states for $\hat{x}$ and $\hat{p}$ are Gaussian wavepackets (coherent states of the harmonic oscillator). They are the quantum states closest to classical particles: a coherent state remains Gaussian under harmonic oscillator evolution, with the center following the classical trajectory while maintaining $\Delta x \cdot \Delta p = \hbar/2$ at all times.
:::

:::{admonition} Discussion
:class: dropdown tip

**Is the uncertainty principle about measurement disturbance or an intrinsic property of quantum states?**

Heisenberg's original argument (1927) used a thought experiment: a gamma-ray microscope that disturbs the electron's momentum when measuring its position. This suggests uncertainty is caused by the act of measurement.

But the Robertson relation $\Delta\hat{A} \cdot \Delta\hat{B} \geq \frac{1}{2}|\langle[\hat{A}, \hat{B}]\rangle|$ is derived purely from the mathematical structure of quantum states — no measurement apparatus is mentioned. The uncertainties $\Delta\hat{A}$ and $\Delta\hat{B}$ refer to the spread of outcomes in repeated measurements on identically prepared ensembles.

- Does this mean uncertainty is a property of the state itself, not of any particular measurement?
- If so, why did Heisenberg's disturbance argument give the right answer?
- Can you design an experiment that distinguishes "intrinsic uncertainty" from "measurement disturbance"? (Hint: look up Ozawa's inequality.)
:::

### Summary

- The expectation value $\langle\hat{O}\rangle = \langle\psi\vert\hat{O}\vert\psi\rangle$ and variance $\mathrm{Var}(\hat{O}) = \langle\hat{O}^2\rangle - \langle\hat{O}\rangle^2$ quantify measurement distributions.
- The commutator $[\hat{A}, \hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}$ measures whether operator order matters.
- Commuting observables share a simultaneous eigenbasis and can be measured together with arbitrary precision.
- The Robertson uncertainty relation $\Delta\hat{A} \cdot \Delta\hat{B} \geq \frac{1}{2}|\langle[\hat{A}, \hat{B}]\rangle|$ is a fundamental constraint on non-commuting observables.
