## Lecture Notes

### Overview

The gauge-invariant Hamiltonian $\hat{H} = \frac{(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2}{2m} + q\Phi$, derived in §4.1.1 purely from local U(1) phase symmetry, contains a profound discovery: it automatically describes charged particles interacting with electromagnetic fields. By computing equations of motion, we extract the Lorentz force and identify the electric and magnetic fields. The gauge principle transforms a mathematical symmetry requirement into the fundamental interaction that shapes the universe.

### Kinetic Momentum

The Heisenberg equation for position gives:

$$

\frac{\mathrm{d}\langle\boldsymbol{x}\rangle}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}\langle[\hat{\boldsymbol{x}}, \hat{H}]\rangle = \frac{\langle\hat{\boldsymbol{p}} - q\boldsymbol{A}\rangle}{m}

$$

using $[\hat{x}_i, \hat{p}_j] = \mathrm{i}\hbar\delta_{ij}$ and $[\hat{x}_i, \boldsymbol{A}(\hat{\boldsymbol{x}})] = 0$.

:::{admonition} Kinetic Momentum
:class: important

$$

\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}, \qquad \boldsymbol{v} = \frac{\hat{\boldsymbol{\pi}}}{m}

$$

The **kinetic momentum** $\hat{\boldsymbol{\pi}}$ determines the particle's velocity. The canonical momentum $\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla$ is gauge-dependent; only $\hat{\boldsymbol{\pi}}$ is directly observable.
:::

:::{admonition} Derivation: Velocity from the Heisenberg Equation
:class: dropdown information

We compute $\frac{\mathrm{d}\langle\hat{\boldsymbol{x}}\rangle}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}\langle[\hat{\boldsymbol{x}}, \hat{H}]\rangle$ with $\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi$.

Since $[\hat{x}_i, q\Phi(\hat{\boldsymbol{x}})] = 0$ (both depend only on $\hat{\boldsymbol{x}}$), only the kinetic term contributes. Using the product rule $[\hat{A}, \hat{B}\hat{C}] = \hat{B}[\hat{A}, \hat{C}] + [\hat{A}, \hat{B}]\hat{C}$:

$$

[\hat{x}_i, (\hat{p}_j - qA_j)(\hat{p}_j - qA_j)] = (\hat{p}_j - qA_j)[\hat{x}_i, \hat{p}_j - qA_j] + [\hat{x}_i, \hat{p}_j - qA_j](\hat{p}_j - qA_j)

$$

Since $[\hat{x}_i, A_j(\hat{\boldsymbol{x}})] = 0$ and $[\hat{x}_i, \hat{p}_j] = \mathrm{i}\hbar\delta_{ij}$:

$$

[\hat{x}_i, \hat{p}_j - qA_j] = \mathrm{i}\hbar\delta_{ij}

$$

Substituting back and summing over $j$:

$$

[\hat{x}_i, (\hat{\boldsymbol{p}} - q\boldsymbol{A})^2] = (\hat{p}_i - qA_i)\,\mathrm{i}\hbar + \mathrm{i}\hbar\,(\hat{p}_i - qA_i) = 2\mathrm{i}\hbar\,\hat{\pi}_i

$$

Therefore:

$$

\frac{\mathrm{d}\langle\hat{x}_i\rangle}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar} \cdot \frac{1}{2m} \cdot 2\mathrm{i}\hbar\langle\hat{\pi}_i\rangle = \frac{\langle\hat{\pi}_i\rangle}{m}

$$

confirming $\boldsymbol{v} = \hat{\boldsymbol{\pi}}/m$. $\checkmark$
:::

:::{admonition} Caution: Kinetic Momentum Does Not Commute
:class: caution

Unlike canonical momentum ($[\hat{p}_i, \hat{p}_j] = 0$), kinetic momentum components satisfy $[\hat{\pi}_i, \hat{\pi}_j] = \mathrm{i}q\hbar\epsilon_{ijk}B_k$ in a magnetic field. This noncommutativity is central to Landau quantization (§4.3).
:::

:::{admonition} Derivation: Kinetic Momentum Commutator
:class: dropdown information

We compute $[\hat{\pi}_i, \hat{\pi}_j]$ where $\hat{\pi}_i = \hat{p}_i - qA_i(\hat{\boldsymbol{x}})$. Since $[\hat{p}_i, \hat{p}_j] = 0$ and $[A_i, A_j] = 0$:

$$

[\hat{\pi}_i, \hat{\pi}_j] = [\hat{p}_i - qA_i,\, \hat{p}_j - qA_j] = -q[\hat{p}_i, A_j] - q[A_i, \hat{p}_j]

$$

Using $[\hat{p}_i, f(\hat{\boldsymbol{x}})] = -\mathrm{i}\hbar\,\partial_i f$:

$$

[\hat{\pi}_i, \hat{\pi}_j] = -q(-\mathrm{i}\hbar\,\partial_i A_j) - q(\mathrm{i}\hbar\,\partial_j A_i) = \mathrm{i}q\hbar(\partial_i A_j - \partial_j A_i)

$$

Recognizing the antisymmetric combination as the curl via the Levi-Civita symbol, $\partial_i A_j - \partial_j A_i = \epsilon_{ijk}B_k$ where $\boldsymbol{B} = \nabla \times \boldsymbol{A}$:

$$

[\hat{\pi}_i, \hat{\pi}_j] = \mathrm{i}q\hbar\,\epsilon_{ijk}B_k

$$

In a uniform magnetic field $\boldsymbol{B} = B\hat{z}$, this gives $[\hat{\pi}_x, \hat{\pi}_y] = \mathrm{i}q\hbar B$, which is the noncommutativity responsible for Landau level quantization (§4.3).
:::

### The Lorentz Force Law

Computing $\frac{\mathrm{d}\langle\hat{\boldsymbol{\pi}}\rangle}{\mathrm{d}t}$ requires the commutator $[\hat{\boldsymbol{p}}, \hat{H}]$. Using $[\hat{p}_i, A_j] = -\mathrm{i}\hbar\partial_i A_j$ and the product rule on $(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2$, one extracts gradient and curl terms from the gauge potentials.

:::{admonition} Lorentz Force from Gauge Theory
:class: important

$$

m\frac{\mathrm{d}\langle\boldsymbol{v}\rangle}{\mathrm{d}t} = q(\langle\boldsymbol{E}\rangle + \langle\boldsymbol{v} \times \boldsymbol{B}\rangle)

$$

where $\boldsymbol{E} = -\nabla\Phi - \frac{\partial\boldsymbol{A}}{\partial t}$ and $\boldsymbol{B} = \nabla \times \boldsymbol{A}$.

This is the **quantum Lorentz force law**, derived entirely from the gauge-invariant Hamiltonian with no assumption about the existence of electromagnetic forces.
:::

:::{admonition} Derivation: Equation of Motion for Kinetic Momentum
:class: dropdown information

**Step 1: Compute $[\hat{\boldsymbol{p}}, \hat{H}]$.** The Hamiltonian is $\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi$. We need two types of commutator.

**(a) $[\hat{p}_i, q\Phi]$:** Since $\Phi = \Phi(\hat{\boldsymbol{x}})$ and $[\hat{p}_i, f(\hat{\boldsymbol{x}})] = -\mathrm{i}\hbar\,\partial_i f$:

$$

[\hat{p}_i, q\Phi] = -\mathrm{i}\hbar\,q\,\partial_i\Phi

$$

**(b) $[\hat{\boldsymbol{p}}, (\hat{\boldsymbol{p}} - q\boldsymbol{A})^2]$:** Using the kinetic momentum commutator $[\hat{\pi}_i, \hat{\pi}_j] = \mathrm{i}q\hbar\epsilon_{ijk}B_k$ and $[\hat{p}_i, \hat{\pi}_j] = [\hat{p}_i, \hat{p}_j - qA_j] = -q[\hat{p}_i, A_j] = \mathrm{i}q\hbar\,\partial_i A_j$, apply the product rule:

$$

[\hat{p}_i, \hat{\pi}_j\hat{\pi}_j] = \hat{\pi}_j[\hat{p}_i, \hat{\pi}_j] + [\hat{p}_i, \hat{\pi}_j]\hat{\pi}_j = \mathrm{i}q\hbar\bigl(\hat{\pi}_j\,\partial_i A_j + (\partial_i A_j)\hat{\pi}_j\bigr)

$$

Combining: $\frac{\mathrm{d}\langle\hat{p}_i\rangle}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}\langle[\hat{p}_i, \hat{H}]\rangle = \frac{q}{2m}\langle\hat{\pi}_j\,\partial_i A_j + (\partial_i A_j)\hat{\pi}_j\rangle - q\langle\partial_i\Phi\rangle$.

**Step 2: From canonical to kinetic momentum.** The kinetic momentum satisfies:

$$

\frac{\mathrm{d}\hat{\pi}_i}{\mathrm{d}t} = \frac{\mathrm{d}\hat{p}_i}{\mathrm{d}t} - q\frac{\partial A_i}{\partial t}

$$

So $m\frac{\mathrm{d}\langle v_i\rangle}{\mathrm{d}t} = \frac{q}{2m}\langle\hat{\pi}_j\,\partial_i A_j + (\partial_i A_j)\hat{\pi}_j\rangle - q\langle\partial_i\Phi\rangle - q\left\langle\frac{\partial A_i}{\partial t}\right\rangle$.

**Step 3: Identify E and v×B.** The last two terms give the electric field: $-q\nabla\Phi - q\partial_t\boldsymbol{A} = q\boldsymbol{E}$. For the magnetic term, symmetrize using $\partial_i A_j = \frac{1}{2}(\partial_i A_j + \partial_j A_i) + \frac{1}{2}(\partial_i A_j - \partial_j A_i)$. The symmetric part contributes $\frac{q}{2m}\langle\hat{\pi}_j\,\partial_j A_i + (\partial_j A_i)\hat{\pi}_j\rangle = q\langle v_j\,\partial_j A_i\rangle$, which combines with $-q\partial_t A_i$ into $-q\frac{\mathrm{d}A_i}{\mathrm{d}t}$, canceling part of the kinetic momentum conversion. The antisymmetric part gives:

$$

\frac{q}{2m}\langle\hat{\pi}_j(\partial_i A_j - \partial_j A_i) + (\partial_i A_j - \partial_j A_i)\hat{\pi}_j\rangle = \frac{q}{2}\langle v_j\epsilon_{ijk}B_k + \epsilon_{ijk}B_k\,v_j\rangle

$$

This is the symmetrized cross product $\frac{q}{2}\langle\boldsymbol{v}\times\boldsymbol{B} - \boldsymbol{B}\times\boldsymbol{v}\rangle$, which equals $q\langle\boldsymbol{v}\times\boldsymbol{B}\rangle$ in the semiclassical limit. Combining all terms yields the quantum Lorentz force:

$$

m\frac{\mathrm{d}\langle\boldsymbol{v}\rangle}{\mathrm{d}t} = q\langle\boldsymbol{E}\rangle + \frac{q}{2}\langle\boldsymbol{v}\times\boldsymbol{B} - \boldsymbol{B}\times\boldsymbol{v}\rangle

$$

The symmetrization $\frac{1}{2}(\boldsymbol{v}\times\boldsymbol{B} - \boldsymbol{B}\times\boldsymbol{v})$ is required because $\hat{\boldsymbol{v}}$ and $\hat{\boldsymbol{B}}$ are operators that generally do not commute. In the semiclassical limit (replacing operators by their expectation values), this reduces to $q(\boldsymbol{E} + \boldsymbol{v}\times\boldsymbol{B})$. $\checkmark$
:::

:::{admonition} Discussion
:class: dropdown tip

In classical mechanics, the Lorentz force $m\boldsymbol{a} = q(\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B})$ is an empirical law. In quantum mechanics, we derived it from the gauge principle alone. Does this mean quantum mechanics is "deeper" than classical mechanics? Could a different gauge group produce a different force law?
:::

### Electric and Magnetic Fields from Potentials

The gauge potentials $(\Phi, \boldsymbol{A})$ are gauge-dependent and not directly observable. The physical fields are:

$$

\boldsymbol{E} = -\nabla\Phi - \frac{\partial\boldsymbol{A}}{\partial t}, \qquad \boldsymbol{B} = \nabla \times \boldsymbol{A}

$$

In spacetime notation with $A_\mu = (\Phi, \boldsymbol{A})$, both fields are encoded in the **field strength tensor**:

:::{admonition} Field Strength Tensor
:class: important

$$

F_{\mu\nu} = \partial_\mu A_\nu - \partial_\nu A_\mu

$$

$F_{\mu\nu}$ is **gauge-invariant**: under $A_\mu \to A_\mu + \partial_\mu\chi$, the antisymmetry gives $F'_{\mu\nu} = F_{\mu\nu}$ because $\partial_\mu\partial_\nu\chi = \partial_\nu\partial_\mu\chi$.
:::

:::{admonition} Derivation: Gauge Invariance of the Field Strength Tensor
:class: dropdown information

Under gauge transformation $A_\mu \to A'_\mu = A_\mu + \partial_\mu\alpha$:

$$

F'_{\mu\nu} = \partial_\mu A'_\nu - \partial_\nu A'_\mu = \partial_\mu(A_\nu + \partial_\nu\alpha) - \partial_\nu(A_\mu + \partial_\mu\alpha)

$$

Expanding:

$$

F'_{\mu\nu} = \partial_\mu A_\nu + \partial_\mu\partial_\nu\alpha - \partial_\nu A_\mu - \partial_\nu\partial_\mu\alpha

$$

Since partial derivatives commute ($\partial_\mu\partial_\nu\alpha = \partial_\nu\partial_\mu\alpha$), the $\alpha$-dependent terms cancel:

$$

F'_{\mu\nu} = \partial_\mu A_\nu - \partial_\nu A_\mu = F_{\mu\nu} \quad\checkmark

$$

This means $\boldsymbol{E}$ and $\boldsymbol{B}$, as components of $F_{\mu\nu}$, are gauge-invariant and physically observable, while $\boldsymbol{A}$ and $\Phi$ individually are not.
:::

### The Gauge Principle Requires Electromagnetism

:::{admonition} The Gauge Principle
:class: important

Starting from only two inputs:

1. Local U(1) phase invariance: $\psi(\boldsymbol{x},t) \to \mathrm{e}^{\mathrm{i}q\chi(\boldsymbol{x},t)}\psi(\boldsymbol{x},t)$
2. The Schrödinger equation retains its form

we were **forced** to introduce a gauge field $\boldsymbol{A}$, replace $\nabla \to \nabla - \mathrm{i}(q/\hbar)\boldsymbol{A}$ (covariant derivative), and adopt $\hat{H} = \frac{(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2}{2m} + q\Phi$. From this, the Lorentz force and the electric/magnetic fields emerged automatically.

**Symmetry $\to$ Interaction**: The gauge field is not freely chosen—it is required by quantum consistency. Every charged particle couples to the same universal gauge field.
:::

:::{admonition} Derivation: Gauge-Invariant Probability Current with Minimal Coupling
:class: dropdown information

The free-particle probability current $\boldsymbol{j}_0 = \frac{\hbar}{2m\mathrm{i}}(\psi^*\nabla\psi - \psi\nabla\psi^*)$ is **not** gauge-invariant: under $\psi \to \mathrm{e}^{\mathrm{i}q\chi/\hbar}\psi$, each gradient picks up a $\nabla\chi$ term. We derive the correct gauge-invariant current from the continuity equation for $\rho = |\psi|^2$.

**Step 1: Derive the continuity equation.** Multiply the minimal-coupling Schrödinger equation by $\psi^*$:

$$
\mathrm{i}\hbar\psi^*\partial_t\psi = \frac{1}{2m}\psi^*(-\mathrm{i}\hbar\nabla - q\boldsymbol{A})^2\psi + q\Phi|\psi|^2
$$

Subtract the complex conjugate (multiply the conjugate equation by $\psi$):

$$
\mathrm{i}\hbar\partial_t|\psi|^2 = \frac{1}{2m}\bigl[\psi^*(-\mathrm{i}\hbar\nabla - q\boldsymbol{A})^2\psi - \psi(\mathrm{i}\hbar\nabla - q\boldsymbol{A})^2\psi^*\bigr]
$$

The scalar potential $q\Phi|\psi|^2$ cancels.

**Step 2: Expand using covariant derivatives.** Let $\hat{\boldsymbol{\pi}} = -\mathrm{i}\hbar\nabla - q\boldsymbol{A}$. The $q^2A^2|\psi|^2$ terms also cancel. Using the vector identity $\psi^*\nabla^2\psi - \psi\nabla^2\psi^* = \nabla\cdot(\psi^*\nabla\psi - \psi\nabla\psi^*)$ and grouping the vector potential terms:

$$
\psi^*\hat{\boldsymbol{\pi}}^2\psi - \psi(\hat{\boldsymbol{\pi}}^*)^2\psi^*
= -\hbar^2\nabla\cdot(\psi^*\nabla\psi - \psi\nabla\psi^*) - 2\mathrm{i}q\hbar\,\nabla\cdot(\boldsymbol{A}|\psi|^2)
$$

where the cross terms combine as $\boldsymbol{A}\cdot\nabla|\psi|^2 + (\nabla\cdot\boldsymbol{A})|\psi|^2 = \nabla\cdot(\boldsymbol{A}|\psi|^2)$.

**Step 3: Identify the current.** Dividing by $\mathrm{i}\hbar$:

$$
\partial_t\rho = -\nabla\cdot\underbrace{\left[\frac{\hbar}{2m\mathrm{i}}(\psi^*\nabla\psi - \psi\nabla\psi^*) + \frac{q}{m}\boldsymbol{A}|\psi|^2\right]}_{\boldsymbol{j}}
$$

So the **gauge-invariant probability current** is:

$$
\boldsymbol{j} = \frac{1}{m}\,\mathrm{Re}\bigl[\psi^*(-\mathrm{i}\hbar\nabla - q\boldsymbol{A})\psi\bigr] = \frac{1}{m}\,\mathrm{Re}[\psi^*\hat{\boldsymbol{\pi}}\psi] = \rho\,\boldsymbol{v}
$$

**Step 4: Verify gauge invariance.** Under $\psi \to \mathrm{e}^{\mathrm{i}q\chi/\hbar}\psi$ and $\boldsymbol{A} \to \boldsymbol{A} + \nabla\chi$:
- The gradient term $\psi^*\nabla\psi$ acquires $+\frac{\mathrm{i}q}{\hbar}|\psi|^2\nabla\chi$ and its conjugate $-\frac{\mathrm{i}q}{\hbar}|\psi|^2\nabla\chi$, contributing $-\frac{q}{m}(\nabla\chi)|\psi|^2$ to $\boldsymbol{j}$.
- The vector potential term $\frac{q}{m}\boldsymbol{A}|\psi|^2$ acquires $+\frac{q}{m}(\nabla\chi)|\psi|^2$.
- These two contributions cancel exactly, leaving $\boldsymbol{j}$ unchanged. $\checkmark$
:::


### Summary

- **Kinetic momentum** $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ determines velocity; its components do not commute in a magnetic field.
- The **Lorentz force** $m\boldsymbol{a} = q(\boldsymbol{E} + \boldsymbol{v}\times\boldsymbol{B})$ is derived from the gauge-invariant Hamiltonian, not assumed.
- **Electric and magnetic fields** arise from gauge potentials: $\boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}$, $\boldsymbol{B} = \nabla\times\boldsymbol{A}$.
- The **field strength tensor** $F_{\mu\nu} = \partial_\mu A_\nu - \partial_\nu A_\mu$ is gauge-invariant.
- The **gauge principle** transforms symmetry into interaction: local U(1) invariance requires electromagnetism.
