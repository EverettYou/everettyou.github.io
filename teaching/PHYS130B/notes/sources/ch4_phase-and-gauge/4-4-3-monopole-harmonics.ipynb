## Lecture Notes

### Overview

When a charged particle moves on a sphere with a magnetic monopole at the center, a key subtlety arises: neither the canonical angular momentum $\hat{\boldsymbol{L}}_{\mathrm{can}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{p}}$ nor the kinetic angular momentum $\hat{\boldsymbol{\Lambda}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{\pi}}$ (where $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$) is separately conserved. The conserved quantity is the **total angular momentum** $\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} - \hbar s\,\boldsymbol{n}$, including the electromagnetic
field's angular momentum $-\hbar s\,\boldsymbol{n}$ (with $\boldsymbol{n}$ the unit radial vector). This $\hat{\boldsymbol{J}}$ satisfies the standard SU(2) algebra $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$, but the minimum eigenvalue is shifted: $j \geq \vert s\vert$ instead of $j \geq 0$. The eigenfunctions are **monopole harmonics** $Y^s_{j,m}(\theta,\varphi)$. The punchline: for $s = 1/2$ (a Dirac monopole), the lowest harmonics form a doublet with $j = 1/2$ — they *are* spin-1/2 spinors. Half-integer angular momentum is not an independent axiom; it emerges from gauge topology.

### Monopole Quantum Number

:::{admonition} Monopole Quantum Number
:class: important

The **monopole quantum number** characterizes the coupling strength:

$$
s = \frac{qg}{4\pi\hbar}
$$ (eq-monopole-qnum)

The Dirac quantization condition $qg = 2\pi n\hbar$ (from the two-patch construction of §4.4.2) restricts $s$ to integer or half-integer values because $s=n/2$:

$$
s = \frac{n}{2} \in \left\{0,\, \pm\tfrac{1}{2},\, \pm 1,\, \pm\tfrac{3}{2},\, \ldots\right\}.
$$

For an electron in the field of a minimal Dirac monopole: $\vert s\vert = 1/2$.
:::

:::{admonition} Convention: Magnetic Charge vs. Monopole Quantum Number
:class: attention

In this section we adopt the flux-defined (Dirac-monopole) convention: the magnetic charge $g$ is identified with the total flux of the monopole, so $\boldsymbol{B} = g\,\boldsymbol{n}/(4\pi r^2)$, with $\boldsymbol{n} = \hat{\boldsymbol{r}}/r$ the unit radial vector ($\hat{\boldsymbol{r}}$ the position operator, $r$ the radial distance), and $\oint\boldsymbol{B}\cdot\mathrm{d}\boldsymbol{S} = g$. The dimensionless quantum number seen by a particle of electric charge $q$ is $s$:

$$
s = \frac{qg}{4\pi\hbar}.
$$

Dirac quantization says $2s\in\mathbb{Z}$. Therefore the **minimal Dirac monopole** is not defined by setting $g=1$ numerically; it is defined by the smallest nonzero allowed coupling, $\vert s\vert=1/2$. Equivalently, a loop enclosing the full solid angle gives Berry/Aharonov-Bohm phase

$$
\begin{split}
\Phi_{\mathrm{Berry}} &= -s\Omega, \\
\Omega = 4\pi &\Rightarrow \Phi_{\mathrm{Berry}} = -4\pi s = -2\pi n.
\end{split}
$$

This is why a minimal monopole has Chern number of magnitude $\vert c_1\vert = 2\vert s\vert = 1$, while a monopole with $s=1$ carries twice the minimal charge.
:::

### Total Angular Momentum

In the monopole background, the kinetic angular momentum $\hat{\boldsymbol{\Lambda}}$ does not commute with itself in the standard way because the gauge field $\boldsymbol{A}$ depends on position. The physically conserved quantity is:

:::{admonition} Total Angular Momentum in a Monopole Field
:class: important

$$
\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} - \hbar s\,\boldsymbol{n}
$$

where $\hat{\boldsymbol{\Lambda}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{\pi}}$ is the kinetic angular momentum (with $\hat{\boldsymbol{r}}$ the position operator), $\boldsymbol{n} = \hat{\boldsymbol{r}}/r$ is the unit radial vector, and $-\hbar s\,\boldsymbol{n}$ is the angular momentum carried by the electromagnetic field.

$\hat{\boldsymbol{J}}$ satisfies the **standard** SU(2) commutation relations:

$$
[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\,\epsilon_{ijk}\hat{J}_k
$$

with eigenvalues:

$$
\begin{split}
\hat{J}^2 \vert s, j, m\rangle &= \hbar^2 j(j+1) \vert s, j, m\rangle, \\
\hat{J}_z \vert s, j, m\rangle &= \hbar m \vert s, j, m\rangle.
\end{split}
$$

**The monopole modifies only the allowed values of $j$** (not $j = 0, 1, 2, \ldots$):

$$
j = \vert s\vert,\; \vert s\vert+1,\; \vert s\vert+2,\; \ldots
$$

:::

The shift $j_{\min} = \vert s\vert$ has a direct physical interpretation: the electromagnetic field already carries angular momentum $-\hbar s\,\boldsymbol{n}$. The total angular momentum cannot be less than this field contribution.

:::{admonition} Derivation: $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$
:class: dropdown information

**Building-block identities.** Two commutators are needed.

*Ingredient 1.* The kinetic angular momentum generates spatial rotations, so it rotates any vector built from position. Applied to the position operator itself, $\hat{\boldsymbol{\Lambda}} = \hat{\boldsymbol{r}}\times\hat{\boldsymbol{\pi}}$ together with $[\hat{r}_a, \hat{\pi}_b] = \mathrm{i}\hbar\delta_{ab}$ gives

$$
[\hat{\Lambda}_i, \hat{r}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{r}_k,
$$

with the gauge potential dropping out because $[\hat{r}_a, \hat{r}_b] = 0$ and $\hat{r}_a$ commutes with $\boldsymbol{A}(\hat{\boldsymbol{r}})$. Because $\hat{\boldsymbol{\Lambda}}$ generates rotations, which preserve the radial distance $r$, it commutes with $r$; the unit radial vector $\boldsymbol{n} = \hat{\boldsymbol{r}}/r$ therefore inherits the same rotation law:

$$
[\hat{\Lambda}_i, n_j] = \mathrm{i}\hbar\epsilon_{ijk}n_k.
$$

*Ingredient 2.* Unlike the field-free case, the kinetic angular momentum does **not** close on itself. In a magnetic field the kinetic momenta fail to commute (§4.1.2):

$$
[\hat{\pi}_i, \hat{\pi}_j] = \mathrm{i}\hbar q\,\epsilon_{ijk}B_k.
$$

For the monopole field $\boldsymbol{B} = g\,\boldsymbol{n}/(4\pi r^2)$ (§4.4.2), restricted to the unit sphere $r = 1$, this radial non-commutativity is set by

$$
q B_k \;\longrightarrow\; \frac{qg}{4\pi}\,n_k \;=\; \hbar s\,n_k,
$$

using the monopole quantum number $s = qg/(4\pi\hbar)$ from {eq}`eq-monopole-qnum`.

To propagate this into the kinetic angular momentum, write $\hat{\Lambda}_i = \epsilon_{iab}\hat{r}_a\hat{\pi}_b$ in index form and expand:

$$
[\hat{\Lambda}_i, \hat{\Lambda}_j] = \epsilon_{iab}\epsilon_{jcd}\,[\hat{r}_a\hat{\pi}_b,\, \hat{r}_c\hat{\pi}_d].
$$

Repeated use of the commutator product rule $[\hat{X}\hat{Y}, \hat{Z}] = \hat{X}[\hat{Y}, \hat{Z}] + [\hat{X}, \hat{Z}]\hat{Y}$ splits the inner commutator into three pieces:

$$
[\hat{r}_a\hat{\pi}_b,\, \hat{r}_c\hat{\pi}_d] = \hat{r}_a\hat{r}_c\,[\hat{\pi}_b, \hat{\pi}_d] + \hat{r}_a\,[\hat{\pi}_b, \hat{r}_c]\,\hat{\pi}_d + \hat{r}_c\,[\hat{r}_a, \hat{\pi}_d]\,\hat{\pi}_b.
$$

The three building-block commutators are the canonical relation from Ingredient 1 in its two index orderings, $[\hat{r}_a, \hat{\pi}_d] = \mathrm{i}\hbar\delta_{ad}$ and $[\hat{\pi}_b, \hat{r}_c] = -\mathrm{i}\hbar\delta_{bc}$, together with the monopole non-commutativity $[\hat{\pi}_b, \hat{\pi}_d] = \mathrm{i}\hbar q\,\epsilon_{bde}B_e$ stated above. Substituting,

$$
[\hat{r}_a\hat{\pi}_b,\, \hat{r}_c\hat{\pi}_d] = \mathrm{i}\hbar q\,\epsilon_{bde}\,\hat{r}_a\hat{r}_c B_e - \mathrm{i}\hbar\,\delta_{bc}\,\hat{r}_a\hat{\pi}_d + \mathrm{i}\hbar\,\delta_{ad}\,\hat{r}_c\hat{\pi}_b.
$$

The last two terms are field-independent. Contracting them with $\epsilon_{iab}\epsilon_{jcd}$ and using $\epsilon_{iab}\epsilon_{jbd} = \delta_{id}\delta_{aj} - \delta_{ij}\delta_{ad}$ and $\epsilon_{iab}\epsilon_{jca} = \delta_{bj}\delta_{ic} - \delta_{bc}\delta_{ij}$:

$$
\begin{split}
-\,\mathrm{i}\hbar\,\epsilon_{iab}\epsilon_{jcd}\,\delta_{bc}\,\hat{r}_a\hat{\pi}_d &= -\,\mathrm{i}\hbar\,(\hat{r}_j\hat{\pi}_i - \delta_{ij}\,\hat{r}_a\hat{\pi}_a), \\
+\,\mathrm{i}\hbar\,\epsilon_{iab}\epsilon_{jcd}\,\delta_{ad}\,\hat{r}_c\hat{\pi}_b &= +\,\mathrm{i}\hbar\,(\hat{r}_i\hat{\pi}_j - \delta_{ij}\,\hat{r}_a\hat{\pi}_a).
\end{split}
$$

The $\delta_{ij}$ pieces cancel in the sum, leaving $\mathrm{i}\hbar\,(\hat{r}_i\hat{\pi}_j - \hat{r}_j\hat{\pi}_i) = \mathrm{i}\hbar\,\epsilon_{ijk}\hat{\Lambda}_k$ — the ordinary rotation algebra.

The first term carries the monopole field. With $q B_e \to \hbar s\,n_e$ and the position operator restricted to the unit sphere ($\hat{r}_a \to n_a$):

$$
\epsilon_{iab}\epsilon_{jcd}\,(\mathrm{i}\hbar q\,\epsilon_{bde}\,\hat{r}_a\hat{r}_c B_e) = \mathrm{i}\hbar^2 s\;\epsilon_{iab}\epsilon_{jcd}\epsilon_{bde}\;n_a n_c n_e.
$$

Contracting $\epsilon_{iab}\epsilon_{bde} = \delta_{id}\delta_{ae} - \delta_{ie}\delta_{ad}$ first:

$$
\begin{split}
\epsilon_{iab}\epsilon_{jcd}\epsilon_{bde}\,n_a n_c n_e &= \epsilon_{jcd}\,(\delta_{id}\delta_{ae} - \delta_{ie}\delta_{ad})\,n_a n_c n_e \\
&= \epsilon_{jci}\,n_c\,(n_e n_e) - \epsilon_{jcd}\,n_c n_d\,n_i \\
&= \epsilon_{ijk}\,n_k,
\end{split}
$$

where $n_e n_e = 1$ on the unit sphere and $\epsilon_{jcd}n_c n_d = 0$ (an antisymmetric symbol contracted with a symmetric product). The field term is therefore $\mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k$.

Adding the field-free and field contributions, the kinetic angular momentum acquires an anomalous term proportional to $\boldsymbol{n}$:

$$
[\hat{\Lambda}_i, \hat{\Lambda}_j] = \mathrm{i}\hbar\epsilon_{ijk}(\hat{\Lambda}_k + \hbar s\,n_k) = \mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k + \mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k.
$$

**Bilinear assembly.** With $\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} - \hbar s\,\boldsymbol{n}$, expand the commutator term by term:

$$
\begin{split}
[\hat{J}_i, \hat{J}_j] &= [\hat{\Lambda}_i - \hbar s\,n_i,\; \hat{\Lambda}_j - \hbar s\,n_j] \\
&= [\hat{\Lambda}_i, \hat{\Lambda}_j] - \hbar s[\hat{\Lambda}_i, n_j] - \hbar s[n_i, \hat{\Lambda}_j] + \hbar^2 s^2[n_i, n_j].
\end{split}
$$

Evaluate each term using ingredients 1 and 2, with $[n_i, \hat{\Lambda}_j] = -[\hat{\Lambda}_j, n_i] = \mathrm{i}\hbar\epsilon_{ijk}n_k$ and $[n_i, n_j] = 0$:

$$
\begin{split}
[\hat{\Lambda}_i, \hat{\Lambda}_j] &= \mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k + \mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k, \\
-\hbar s\,[\hat{\Lambda}_i, n_j] &= -\,\mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k, \\
-\hbar s\,[n_i, \hat{\Lambda}_j] &= -\,\mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k, \\
\hbar^2 s^2\,[n_i, n_j] &= 0.
\end{split}
$$

**Cancellation.** The anomalous $+\mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k$ from ingredient 2 and the two cross terms $-\mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k$ sum to $-\mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k$. Adding this to $\mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k$ reassembles $\hat{J}_k = \hat{\Lambda}_k - \hbar s\,n_k$:

$$
[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k - \mathrm{i}\hbar^2 s\,\epsilon_{ijk}n_k = \mathrm{i}\hbar\epsilon_{ijk}(\hat{\Lambda}_k - \hbar s\,n_k) = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k.
$$

The field contribution $-\hbar s\,\boldsymbol{n}$ is exactly what is needed to absorb the anomalous term and restore the standard SU(2) algebra.

:::

### Monopole Harmonics

The eigenfunctions of $\hat{J}^2$ and $\hat{J}_z$ in the monopole background are **monopole harmonics**:

:::{admonition} Monopole Harmonics
:class: important

$Y^s_{j,m}(\theta,\varphi)$ are simultaneous eigenfunctions of $\hat{J}^2$ and $\hat{J}_z$:

$$
\begin{split}
\hat{J}^2 Y^s_{j,m} &= \hbar^2 j(j+1) Y^s_{j,m}, \\
\hat{J}_z Y^s_{j,m} &= \hbar m Y^s_{j,m}.
\end{split}
$$

with $j = \vert s\vert, \vert s\vert+1, \ldots$ and $m = -j, \ldots, j$.

**Orthonormality:** $\displaystyle\int \sin\theta\,\mathrm{d}\theta\,\mathrm{d}\varphi\,(Y^s_{j,m})^* Y^s_{j',m'} = \delta_{jj'}\delta_{mm'}$

**Reduction:** For $s = 0$, $Y^0_{l,m} = Y_{lm}(\theta,\varphi)$ (ordinary spherical harmonics).
:::

The explicit form of $Y^s_{j,m}$ requires the two-patch construction (§4.4.2): the monopole gauge field cannot be described by a single vector potential on the whole sphere, so northern and southern hemispheres use separate gauges, related by a transition function on the equator.

Summary of how the monopole modifies the spectrum:

| $s$ | $j_{\min}$ | Degeneracy at $j_{\min}$ |
|-----|-----------|--------------------------|
| $0$ (no monopole) | $0$ | $1$ (singlet) |
| $1/2$ (Dirac) | $1/2$ | $2$ (doublet = spin-1/2) |
| $1$ | $1$ | $3$ (triplet) |
| $3/2$ | $3/2$ | $4$ (quartet) |

### Punchline: Spin-1/2 from Monopole Topology

:::{admonition} Spin-1/2 Emerges from $s = 1/2$
:class: important

For a Dirac monopole ($\vert s\vert = 1/2$), the minimum angular momentum is $j_{\min} = 1/2$. The lowest states are:

$$
\left\{\, \vert \tfrac{1}{2}, +\tfrac{1}{2}\rangle,\ \vert \tfrac{1}{2}, -\tfrac{1}{2}\rangle \,\right\}
$$

These form a **doublet** transforming exactly as spin-1/2: under a $2\pi$ rotation, each state acquires a factor $\mathrm{e}^{2\pi\mathrm{i}s} = \mathrm{e}^{\pi\mathrm{i}} = -1$ (the spinor sign flip).

**Conclusion**: Half-integer angular momentum is **not** an independent axiom. It is the minimum angular momentum enforced by the topology of the monopole gauge bundle.
:::

### Connection to Berry Phase

The monopole structure of angular momentum has a direct physical manifestation through Berry phase (§4.2). Consider a spin-1/2 particle in a slowly varying magnetic field, and let $\boldsymbol{R}$ denote the field vector regarded as a point in parameter space — the two spin eigenstates are degenerate at the origin $\boldsymbol{R} = 0$. As $\boldsymbol{R}$ traces a closed loop, the spin acquires a geometric phase $\Phi_{\mathrm{Berry}} = -\frac{1}{2}\Omega$, where $\Omega$ is the solid angle the loop subtends at the origin. The corresponding Berry curvature is the radial monopole field

$$
\boldsymbol{F}(\boldsymbol{R}) = -\frac{1}{2}\,\frac{\hat{\boldsymbol{R}}}{R^2},
$$

a monopole of strength $-1/2$ at the degeneracy point $\boldsymbol{R} = 0$, with $\hat{\boldsymbol{R}}$ the unit vector and $R = \vert\boldsymbol{R}\vert$. The $1/R^2$ falloff makes the curvature flux the same through *every* closed surface enclosing the origin — the topological hallmark of a monopole. Restricted to the unit sphere of field directions ($R = 1$), the curvature reduces to $\boldsymbol{F} = -\tfrac{1}{2}\hat{\boldsymbol{R}}$, exactly the spin-1/2 Berry curvature derived in §4.2.1; its flux through the enclosed cap reproduces $\Phi_{\mathrm{Berry}} = -\frac{1}{2}\Omega$ by Stokes' theorem.

:::{admonition} Spin = Monopole in Parameter Space
:class: important

The Berry curvature of a spin-$s$ system is the field of a **monopole** at the origin of parameter space — of strength $-s$ for the aligned eigenstate (§4.2.1 finds strength $-1/2$ for spin-1/2; the sign depends on which eigenstate is transported, the magnitude does not). This is not an analogy — it is the same mathematical structure.

The Chern number — the curvature flux through any closed surface enclosing the degeneracy, divided by $2\pi$ — is the integer topological invariant

$$
\vert c_1\vert = \frac{1}{2\pi}\left\vert\,\oint \boldsymbol{F}\cdot\mathrm{d}\boldsymbol{S}\,\right\vert = 2\vert s\vert
$$

This closes the circle: spin-1/2 emerges from monopole topology (§4.4.2), and the Berry phase of spin-1/2 is itself a monopole field (§4.2.1).
:::

### Summary

- The **conserved total angular momentum** in a monopole field is $\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} - \hbar s\,\boldsymbol{n}$, satisfying standard SU(2): $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$.

- **Monopole harmonics** $Y^s_{j,m}$ are eigenstates of $\hat{J}^2$ and $\hat{J}_z$ with eigenvalues $\hbar^2 j(j+1)$, but $j_{\min} = \vert s\vert$ instead of 0.

- The **shift $j_{\min} = \vert s\vert$** reflects the angular momentum already carried by the electromagnetic field.

- For the **Dirac monopole** ($s = 1/2$), the lowest harmonics form a spin-1/2 doublet — half-integer angular momentum emerges from topology, not as an axiom.

- The **Berry curvature** of spin-$s$ is a monopole of strength $-s$ in parameter space — spin is topological.
