## Lecture Notes

The gauge field $\boldsymbol{A}$ and the field strength tensor $F_{ij}$ represent two complementary aspects of gauge theory: the connection (which is gauge-dependent and unobservable in isolation) and the curvature (which is gauge-invariant and represents the physical field). Understanding their relationship is the bridge between abstract differential geometry and concrete physics.

### The Gauge Connection (Vector Potential)

#### Definition and Interpretation

The gauge field $\boldsymbol{A}(\boldsymbol{r}, t)$ appears in the covariant derivative:

$$


D_i = \partial_i + \mathrm{i}qA_i

$$


This field is called the **gauge connection** or **vector potential**. In electromagnetism, $\boldsymbol{A}$ is the familiar vector potential related to the magnetic field $\boldsymbol{B} = \nabla \times \boldsymbol{A}$.

:::{admonition} Gauge Connection (Vector Potential)
:class: important

The **gauge connection** $A_i(\boldsymbol{r}, t)$ is the component of a gauge field that couples to charges according to the minimal coupling prescription:

$$


(\partial_i - \mathrm{i}qA_i)\psi = D_i\psi

$$


The connection is **gauge-dependent**: under a gauge transformation $\psi \to \mathrm{e}^{\mathrm{i}q\chi}\psi$, the connection transforms as:

$$


A_i \to A_i - \partial_i\chi

$$


Because the connection transforms, it is **not directly observable**. Only gauge-invariant combinations (like the field strength) have direct physical meaning.
:::

### Field Strength Tensor: The Commutator of Covariant Derivatives

The key insight is that the field strength—the true, gauge-invariant physical field—can be extracted from the commutator of two covariant derivatives.

#### Derivation

Consider the commutator of covariant derivatives acting on a scalar wavefunction $\psi$:

$$


[D_i, D_j]\psi = [(\partial_i + \mathrm{i}qA_i), (\partial_j + \mathrm{i}qA_j)]\psi

$$


Expand carefully:

$$


[D_i, D_j]\psi = D_i(D_j\psi) - D_j(D_i\psi)

$$


$$

= (\partial_i + \mathrm{i}qA_i)[(\partial_j + \mathrm{i}qA_j)\psi] - (\partial_j + \mathrm{i}qA_j)[(\partial_i + \mathrm{i}qA_i)\psi]

$$


$$

= \partial_i\partial_j\psi + \mathrm{i}qA_i\partial_j\psi + \mathrm{i}q(\partial_i A_j)\psi + \mathrm{i}qA_j(\partial_i\psi) - (i \leftrightarrow j)

$$


The $\partial_i\partial_j\psi$ and $\partial_i\partial_j\psi$ terms cancel (because mixed partials commute for smooth functions). Collecting terms:

$$


[D_i, D_j]\psi = \mathrm{i}q[(\partial_i A_j) - (\partial_j A_i)]\psi + \mathrm{i}q[A_i\partial_j - A_j\partial_i]\psi

$$


The second bracket vanishes because $[A_i, \partial_j] = 0$ (the vector potential commutes with spatial derivatives). Therefore:

$$


[D_i, D_j]\psi = \mathrm{i}q(\partial_i A_j - \partial_j A_i)\psi

$$


:::{admonition} Field Strength Tensor
:class: important

Define the **field strength tensor** as:

$$


F_{ij} = \partial_i A_j - \partial_j A_i

$$


Then:

$$


[D_i, D_j] = \mathrm{i}qF_{ij}

$$


The field strength tensor is **gauge-invariant**: under a gauge transformation, $F_{ij}' = F_{ij}$. This is because:

$$


F_{ij}' = \partial_i A_j' - \partial_j A_i' = \partial_i(A_j - \partial_j\chi) - \partial_j(A_i - \partial_i\chi) = F_{ij}

$$

(The derivatives of $\chi$ cancel.)

:::

#:::{admonition} Derivation: Gauge Invariance of Field Strength
:class: dropdown

Under gauge transformation $\psi \to \mathrm{e}^{\mathrm{i}q\chi}\psi$ and $A_i \to A_i - \partial_i\chi$:

$$


F_{ij}' = \partial_i(A_j - \partial_j\chi) - \partial_j(A_i - \partial_i\chi)

$$


$$

= \partial_i A_j - \partial_i\partial_j\chi - \partial_j A_i + \partial_j\partial_i\chi

$$


$$

= (\partial_i A_j - \partial_j A_i) + (\partial_j\partial_i\chi - \partial_i\partial_j\chi) = F_{ij} + 0

$$


since mixed partial derivatives are equal for smooth functions. Thus $F_{ij}$ is gauge-invariant.
:::

### Electromagnetic Field Strength

In electromagnetism, the field strength tensor is a 4×4 matrix where $i, j \in \{0, 1, 2, 3\}$ (spacetime indices). With $A_0 = \Phi/c$ (scalar potential in spacetime units), the components are:

$$


F_{0i} = \partial_0 A_i - \partial_i A_0 = -\frac{1}{c}\partial_t A_i - \partial_i\Phi

$$


relates to the electric field. The spatial components:

$$


F_{ij} = \partial_i A_j - \partial_j A_i

$$


relate to the magnetic field via $B_k = \frac{1}{2}\epsilon_{ijk}F_{ij}$.

For a 3D system with scalar potential $\Phi$ and vector potential $\boldsymbol{A}$:

:::{admonition} Example: Electromagnetic Fields from Potentials
:class: example

The electric field is:

$$


\boldsymbol{E} = -\nabla\Phi - \frac{\partial\boldsymbol{A}}{\partial t}

$$


The magnetic field is:

$$


\boldsymbol{B} = \nabla \times \boldsymbol{A}

$$


These are obtained from components of the field strength tensor. Both are gauge-invariant: different gauge choices for $(\Phi, \boldsymbol{A})$ produce the same $(\boldsymbol{E}, \boldsymbol{B})$.
:::

### Parallel Transport and Holonomy

One of the deepest insights connecting geometry to physics is the notion of **parallel transport** along a path.

#### Concept

Imagine a quantum state "moving" along a path in space while interacting with a gauge field. The state's phase is "parallel transported" along the path, meaning its phase evolution is determined by the local connection.

:::{admonition} Discussion
:class: dropdown tip

In a different gauge, where $A_i' = A_i - \partial_i\chi$, how does the Wilson line $U[\mathcal{C}]$ change? If we measure the same phase shift with a different gauge choice, how can we reconcile this with gauge invariance? What physical quantity remains invariant?
:::

For a particle moving along a path $\mathcal{C}$ parametrized by $s$, with $x_i(s)$ its position, the **path-ordered exponential** (or **Wilson line**) is:

$$


U[\mathcal{C}] = \mathcal{P}\exp\left(-\mathrm{i}q\int_{\mathcal{C}} A_i(\boldsymbol{r}(s))\,\frac{\mathrm{d}x_i}{\mathrm{d}s}\,\mathrm{d}s\right)

$$


This describes how a quantum state is rotated as it travels along the path $\mathcal{C}$.

#### Phase Circulation Around a Loop

Consider a closed loop $\mathcal{C}$. The phase accumulated when a state is parallel-transported around the loop is:

$$


\phi_{\text{loop}} = q\oint_{\mathcal{C}} A_i(\boldsymbol{r})\,\mathrm{d}x_i

$$


By Stokes' theorem:

$$


\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{r} = \int_{\text{surface}} (\nabla \times \boldsymbol{A}) \cdot \mathrm{d}\boldsymbol{S} = \int_{\text{surface}} \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S} = \Phi_B

$$


where $\Phi_B$ is the magnetic flux through the surface bounded by the loop. Therefore:

:::{admonition} Holonomy and Flux
:class: important

The phase accumulated by parallel transport around a closed loop equals the flux of the field strength through the enclosed surface:

$$


\phi_{\text{loop}} = q\int_{\text{surface}} F \cdot \mathrm{d}S

$$


This is a **topological** result: it depends only on the integral of the field strength, not on the detailed shape of the loop.
:::

:::{admonition} Discussion
:class: dropdown tip

The field strength tensor $F_{ij}$ is defined as a commutator of covariant derivatives. In ordinary calculus, the commutator of $\partial_i$ and $\partial_j$ vanishes. Why does adding the gauge field make the covariant derivatives non-commuting? What is the geometric meaning: does this non-commutativity measure something about space itself?
:::

### Non-Abelian Gauge Fields (Brief Introduction)

For more complicated gauge groups (like SU(3) in quantum chromodynamics), the gauge field components themselves are matrices. The field strength becomes:

$$


F_{ij} = \partial_i A_j - \partial_j A_i + \mathrm{i}q[A_i, A_j]

$$


The commutator term makes non-abelian gauge theories **nonlinear**: the field strength depends on the fields themselves, creating self-interactions that are absent in electromagnetism.

### Summary

- The **gauge connection** $A_i$ couples to charges via minimal coupling but is gauge-dependent and unobservable in isolation.
- The **field strength tensor** $F_{ij} = \partial_i A_j - \partial_j A_i$ is obtained from the commutator of covariant derivatives and is gauge-invariant.
- In electromagnetism, $F_{ij}$ encodes the electric and magnetic fields.
- **Parallel transport** along a path is described by the Wilson line; phase accumulated around a closed loop equals the flux of the field strength through the surface.
- Non-abelian gauge theories have self-interacting fields due to $[A_i, A_j]$ terms in the field strength.
