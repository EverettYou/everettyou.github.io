## Lecture Notes

### Overview

The Aharonov-Bohm (AB) effect demonstrates that a charged particle's quantum phase depends on the electromagnetic vector potential $\boldsymbol{A}$, even in regions where the magnetic field vanishes. This reveals that the gauge potential is physically real in quantum mechanics—not merely a mathematical convenience. The effect leads directly to **flux quantization**: in systems with a macroscopic wavefunction (superconductors), magnetic flux is restricted to discrete quantum units $\Phi = n\Phi_0$.

The AB phase is a special case of the Berry phase (§4.2.1): the parameter space is the position of the electron around the solenoid. As the electron's position winds around the solenoid, the vector potential traces a closed loop in parameter space, accumulating geometric phase.

### The Aharonov-Bohm Setup

A solenoid produces a uniform magnetic field $\boldsymbol{B}$ inside but zero field outside. An electron beam splits around the solenoid—one path above, one below—then recombines at a screen.

**Classical prediction:** No effect. The electron experiences no Lorentz force ($\boldsymbol{F} = q\boldsymbol{v} \times \boldsymbol{B} = 0$) in the field-free region, so both paths are identical.

**Quantum reality:** The interference pattern *shifts*. The two beams acquire different quantum phases, even though neither experiences a magnetic force.

Schematic AB setup: paths $C_\mathrm{upper}$ and $C_\mathrm{lower}$ enclose flux; $\boldsymbol{B}$ is confined to the solenoid, but $\boldsymbol{A}$ extends outside, so the relative phase depends on $\oint \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l}$.

For a solenoid of radius $R$ with internal field $B$ along $\hat{\boldsymbol{z}}$, the vector potential outside ($r > R$) in cylindrical coordinates is

$$
\boldsymbol{A}_\text{out} = \frac{\Phi}{2\pi r}\hat{\boldsymbol{\varphi}}
$$

where $\Phi = \pi R^2 B$ is the enclosed flux. The curl vanishes ($\nabla \times \boldsymbol{A} = 0$), yet $\boldsymbol{A} \neq 0$. This is possible because the region is not simply connected—$\boldsymbol{A}$ cannot be written as a gradient globally.

### Two-Path Interference

In a $\boldsymbol{B}=0$ region, a charged particle traversing a path $\mathcal{C}$ accumulates a single-path phase $\Phi_{\mathrm{path}}[\mathcal{C}] = (q/\hbar)\int_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$, obtained by gauging $\boldsymbol{A}$ away locally with the gauge transformation of [§4.1.3 Gauge Invariance](4-1-3-gauge-invariance) (derivation below).

:::{admonition} Derivation: Single-path phase from local gauge transformation
:class: dropdown information

**Setup.** In any simply connected sub-region where $\boldsymbol{B} = \nabla \times \boldsymbol{A} = 0$, the vector potential is curl-free and can be written as a pure gradient. Along a chosen path $\mathcal{C}$ from $\boldsymbol{x}_0$ to $\boldsymbol{x}$, define the gauge function

$$
\alpha(\boldsymbol{x}) \equiv -\int_{\boldsymbol{x}_0,\,\mathcal{C}}^{\boldsymbol{x}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l}'.
$$

Path-independence of line integrals of a curl-free field within the simply connected region (Stokes' theorem with $\nabla\times\boldsymbol{A}=0$) makes $\alpha$ a well-defined function of the endpoint, and the fundamental theorem of calculus gives $\nabla\alpha = -\boldsymbol{A}$.

**Apply the local gauge transformation** from [§4.1.3](4-1-3-gauge-invariance) § *Gauge Transformations* with this $\alpha$:

$$
\begin{split}
\boldsymbol{A} &\to \boldsymbol{A} + \nabla\alpha = 0, \\
\psi &\to \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi.
\end{split}
$$

The transformed wavefunction $\tilde\psi(\boldsymbol{x}) \equiv \mathrm{e}^{\mathrm{i}q\alpha(\boldsymbol{x})/\hbar}\psi(\boldsymbol{x})$ satisfies the Schrödinger equation in the $\boldsymbol{A}=0$ gauge along $\mathcal{C}$ and therefore carries *no* path-dependent EM phase.

**Invert the transformation** to read off the path-dependent phase of the original $\psi$. Using $\alpha(\boldsymbol{x}) = -\int_{\mathcal{C}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l}'$,

$$
\psi(\boldsymbol{x}) = \mathrm{e}^{-\mathrm{i}q\alpha(\boldsymbol{x})/\hbar}\,\tilde\psi(\boldsymbol{x}) = \tilde\psi(\boldsymbol{x})\,\exp\!\left(\frac{\mathrm{i}q}{\hbar}\int_{\mathcal{C}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l}'\right),
$$

so along $\mathcal{C}$ the wavefunction differs from its $\boldsymbol{A}=0$ counterpart by the path-dependent phase

$$
\Phi_{\mathrm{path}}[\mathcal{C}] = \frac{q}{\hbar}\int_{\mathcal{C}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l}.
$$

**Connection to Berry phase.** This is the line-integral Berry-phase form developed in [§4.2.1 Berry Phase](4-2-1-berry-phase), with the EM gauge connection $q\boldsymbol{A}/\hbar$ playing the role of the Berry connection $\boldsymbol{A}_n$ (the identification noted there). Around the solenoid the field-free region is *multiply* connected, so a global $\boldsymbol{A}=0$ gauge does not exist and the two interfering paths' phases need not agree — that mismatch is the AB phase derived next.
:::

The two interfering paths acquire different phases, so their relative phase is the difference of two line integrals:

$$
\begin{split}
\Delta\Phi_{\mathrm{AB}}
  &= \Phi_{\mathrm{path}}[\mathcal{C}_{\mathrm{upper}}] - \Phi_{\mathrm{path}}[\mathcal{C}_{\mathrm{lower}}] \\
  &= \frac{q}{\hbar}\int_{\mathcal{C}_{\mathrm{upper}}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}
   - \frac{q}{\hbar}\int_{\mathcal{C}_{\mathrm{lower}}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}.
\end{split}
$$

Reversing the orientation of the lower path flips the sign of its line integral and joins the two segments into a single closed contour $\mathcal{C} = \mathcal{C}_{\mathrm{upper}} - \mathcal{C}_{\mathrm{lower}}$ that encircles the solenoid:

$$
\Delta\Phi_{\mathrm{AB}} = \frac{q}{\hbar}\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}.
$$

Stokes' theorem turns the loop integral into a flux through any surface $\mathcal{S}$ with $\partial\mathcal{S} = \mathcal{C}$, and $\boldsymbol{B} = \nabla \times \boldsymbol{A}$ identifies the integrand as the magnetic field:

$$
\begin{split}
\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}
  &= \int_{\mathcal{S}} (\nabla \times \boldsymbol{A}) \cdot \mathrm{d}\boldsymbol{S} \\
  &= \int_{\mathcal{S}} \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S} \\
  &= \Phi,
\end{split}
$$

where $\Phi$ is the total magnetic flux enclosed by the loop. Even though the trajectory itself stays in the field-free region ($\boldsymbol{B} = 0$ along $\mathcal{C}$), any cap $\mathcal{S}$ spanning the loop must cut the solenoid and capture the full flux $\Phi = \pi R^2 B$. Substituting back gives the Aharonov-Bohm phase

$$
\Delta\Phi_{\mathrm{AB}} = \frac{q\Phi}{\hbar},
$$

so the interference pattern shifts by $\Delta\Phi_{\mathrm{AB}}/(2\pi)$ of a fringe spacing as the enclosed flux is varied.

:::{admonition} Aharonov-Bohm Phase Shift
:class: important

A charged particle traveling around a solenoid acquires a relative phase

$$
\Delta\Phi_{\mathrm{AB}} = \frac{q}{\hbar}\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q\Phi}{\hbar}
$$

between the two interfering paths, where $\Phi$ is the magnetic flux enclosed by the loop. This shift occurs despite $\boldsymbol{B} = 0$ along the trajectory itself.
:::

:::{admonition} Topological Invariance
:class: important

The AB phase $\Delta\Phi_{\mathrm{AB}} = q\Phi/\hbar$ depends only on whether the loop encircles the solenoid (the winding number), not on its shape or size. Two loops that wind the same number of times around the solenoid produce the same phase shift; smoothly deforming a loop without crossing the solenoid leaves $\Delta\Phi_{\mathrm{AB}}$ invariant.
:::

:::{admonition} Discussion: is $\boldsymbol{A}$ more fundamental than $\boldsymbol{B}$?
:class: dropdown tip

In classical mechanics, a particle's trajectory depends only on $\boldsymbol{E}$ and $\boldsymbol{B}$ (the Lorentz force). The vector potential $\boldsymbol{A}$ does not appear. In quantum mechanics, $\boldsymbol{A}$ enters the Schrödinger equation via minimal coupling, making it observable in gauge-invariant combinations. Is the AB effect a sign that quantum mechanics is "deeper" than classical mechanics, or simply that we must describe particles by wavefunctions rather than trajectories?
:::

### Gauge Invariance and Physical Meaning

Under a gauge transformation $\boldsymbol{A} \to \boldsymbol{A} + \nabla\alpha$, $\psi \to \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi$, the phase on a single path changes by $(q/\hbar)[\alpha(\text{end}) - \alpha(\text{start})]$. But the *relative phase* between two interfering paths sharing the same endpoints is gauge-invariant:

$$
\Delta\Phi_{\mathrm{AB}}' = \Phi_{\mathrm{path},1}' - \Phi_{\mathrm{path},2}' = \Delta\Phi_{\mathrm{AB}}
$$

:::{admonition} The Observable is Relative Phase
:class: attention

The **relative phase difference** between two paths is gauge-invariant and physically measurable. The absolute phase on a single path is gauge-dependent and unobservable. The circulation $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \Phi$ is gauge-invariant because it equals the enclosed flux $\int \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S}$.
:::

:::{admonition} The Vector Potential is Physical in QM
:class: attention

Classical electromagnetism treats $\boldsymbol{A}$ as a mathematical convenience—only $\boldsymbol{E}$ and $\boldsymbol{B}$ are "real." The AB effect demonstrates that in quantum mechanics, gauge-invariant combinations of $\boldsymbol{A}$ (like $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$) have direct observable consequences. This does not contradict gauge invariance; it shows that $\boldsymbol{A}$ encodes physical information that $\boldsymbol{B}$ alone does not capture in multiply connected regions.
:::

### Flux Quantization

In a superconductor, the macroscopic order parameter is a complex field $\Psi(\boldsymbol{r}) = \vert\Psi\vert\,\mathrm{e}^{\mathrm{i}\Theta(\boldsymbol{r})}$ describing the Cooper-pair condensate (charge $q = 2e$). Single-valuedness of $\Psi$ requires that the phase $\Theta$ winds by an integer multiple of $2\pi$ around any closed loop in the bulk:

$$
\oint \nabla\Theta \cdot \mathrm{d}\boldsymbol{l} = 2\pi n, \qquad n \in \mathbb{Z}.
$$

Deep inside the superconductor (beyond the London penetration depth) the supercurrent vanishes. The condensate carries the charge supercurrent

$$
\boldsymbol{j}_s = \frac{q\vert\Psi\vert^2}{m}\bigl(\hbar\nabla\Theta - q\boldsymbol{A}\bigr),
$$

obtained from the gauge-coupled probability current $\boldsymbol{j} = (1/m)\mathrm{Re}[\Psi^*\hat{\boldsymbol{\pi}}\Psi]$ for the kinetic-momentum operator $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ of [§4.1.2 Electromagnetic Coupling](4-1-2-electromagnetic-coupling) (derivation below). Setting $\boldsymbol{j}_s = 0$ in the bulk and using $\vert\Psi\vert \neq 0$ then yields the *rigid-condensate* identification of the phase gradient with the vector potential:

$$
\begin{split}
\boldsymbol{j}_s &= \frac{q\vert\Psi\vert^2}{m}\bigl(\hbar\nabla\Theta - q\boldsymbol{A}\bigr) = 0 \\
\Longrightarrow\quad \hbar\nabla\Theta - q\boldsymbol{A} &= 0 \\
\Longrightarrow\quad \nabla\Theta &= \frac{q}{\hbar}\,\boldsymbol{A}.
\end{split}
$$

:::{admonition} Derivation: Supercurrent for a Cooper-pair condensate
:class: dropdown information

Starting from the gauge-coupled probability current $\boldsymbol{j} = (1/m)\mathrm{Re}[\Psi^*\hat{\boldsymbol{\pi}}\Psi]$ and the kinetic-momentum operator $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A} = -\mathrm{i}\hbar\nabla - q\boldsymbol{A}$, write the macroscopic wavefunction in polar form $\Psi = \vert\Psi\vert\,\mathrm{e}^{\mathrm{i}\Theta}$ with $\vert\Psi\vert$ and $\Theta$ real. Evaluating $\hat{\boldsymbol{\pi}}\Psi$ by the product rule on the exponential,

$$
\begin{split}
\hat{\boldsymbol{\pi}}\Psi
  &= -\mathrm{i}\hbar\,\nabla\bigl(\vert\Psi\vert\,\mathrm{e}^{\mathrm{i}\Theta}\bigr)
     - q\boldsymbol{A}\,\vert\Psi\vert\,\mathrm{e}^{\mathrm{i}\Theta} \\
  &= \bigl(-\mathrm{i}\hbar\,\nabla\vert\Psi\vert
     + \hbar\vert\Psi\vert\,\nabla\Theta
     - q\boldsymbol{A}\,\vert\Psi\vert\bigr)\,\mathrm{e}^{\mathrm{i}\Theta}.
\end{split}
$$

Multiplying by $\Psi^* = \vert\Psi\vert\,\mathrm{e}^{-\mathrm{i}\Theta}$ collapses the exponential:

$$
\Psi^*\hat{\boldsymbol{\pi}}\Psi
  = -\mathrm{i}\hbar\,\vert\Psi\vert\,\nabla\vert\Psi\vert
     + \vert\Psi\vert^2\bigl(\hbar\nabla\Theta - q\boldsymbol{A}\bigr).
$$

The first term is purely imaginary ($\vert\Psi\vert$ and $\nabla\vert\Psi\vert$ are both real); the second is purely real. Taking the real part and dividing by $m$ gives the probability current

$$
\boldsymbol{j} = \frac{1}{m}\mathrm{Re}[\Psi^*\hat{\boldsymbol{\pi}}\Psi]
  = \frac{\vert\Psi\vert^2}{m}\bigl(\hbar\nabla\Theta - q\boldsymbol{A}\bigr).
$$

The charge supercurrent is $\boldsymbol{j}_s = q\boldsymbol{j}$ (one factor of $q$ converts probability density to charge density), reproducing the formula used above.
:::

Substituting this into the single-valuedness condition and using $\oint \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l} = \Phi$ for the enclosed flux gives

$$
\begin{split}
\oint \nabla\Theta \cdot \mathrm{d}\boldsymbol{l}
  &= \frac{q}{\hbar}\oint \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{l}
   = \frac{q\Phi}{\hbar} = 2\pi n \\
\Longrightarrow \Phi
  &= n\,\frac{h}{q} = n\,\Phi_0,
\end{split}
$$

where $q = 2e$ for Cooper pairs and $\Phi_0 = h/(2e)$ is the superconducting flux quantum.

:::{admonition} Flux Quantum (General Definition)
:class: important

$$
\boxed{\Phi_0 = \frac{h}{q}}
$$

For any coherent charged degree of freedom with effective charge $q$, the enclosed flux enters interference through the dimensionless ratio $\Phi/\Phi_0$.
:::

:::{admonition} Examples: how charge sets $\Phi_0$
:class: example

- **Superconducting loops (SQUID context):** the coherent carriers are Cooper pairs, so $q=2e$ and $\Phi_0=h/(2e)$. This is the superconducting flux quantum (used operationally in SQUIDs; see **[4.2.3 Flux Ring](4-2-3-flux-ring)**). The measured value $h/(2e)$ (Doll & Näbauer; Deaver & Fairbank, 1961) was early evidence for Cooper pairing.

- **Normal metallic systems (including graphene):** the relevant coherent charge is $q=e$, so $\Phi_0=h/e$. This electron flux quantum appears in AB rings and is the natural flux scale in quantum Hall settings (see **[4.3.3 Quantum Hall Effect](4-3-3-quantum-hall-effect)**).
:::

:::{admonition} Discussion: flux quantization in superconductors
:class: dropdown tip

Flux quantization in a superconductor requires two ingredients: (1) a macroscopic wavefunction with single-valuedness, and (2) Cooper pairing that sets the charge to $2e$. Does knowing the flux quantum $\Phi_0 = h/(2e)$ tell us that pairing *must* occur, or only that paired carriers exist? Could a system have flux quantization with $\Phi_0 = h/e$?

Suppose an experiment reports a dominant period consistent with $\Phi_0 = h/(4e)$. What are plausible explanations? Distinguish carefully between (i) genuinely different coherent charge quanta and (ii) effective period halving from junction geometry, higher harmonics in current-phase relations, or multi-path interference.
:::

### Experimental Confirmation

**Aharonov-Bohm effect:** Tonomura et al. (1986) used electron holography on tiny solenoids ($\sim$500 nm diameter) to directly observe interference pattern shifts as solenoid current varied, in perfect agreement with theory.

**Flux quantization:** Doll & Näbauer (1961) and Deaver & Fairbank (1961) measured the superconducting value $\Phi_0 = h/(2e)$ in tin cylinders. Modern SQUIDs routinely exploit flux quantization with extraordinary precision (see [§4.2.3 Flux Ring](4-2-3-flux-ring)).

### Summary

- The **Aharonov-Bohm phase** $\Delta\Phi_{\mathrm{AB}} = q\Phi/\hbar$ arises from the vector potential in field-free regions.

- The **relative phase** between two paths is gauge-invariant and observable through interference.

- The vector potential has **physical reality** in quantum mechanics: gauge-invariant combinations of $\boldsymbol{A}$ are directly measurable.

- **Flux quantization** follows from single-valuedness: $\Phi = n\Phi_0$ with the general definition $\Phi_0=h/q$ (superconductors: $q=2e$; normal electrons: $q=e$).

- These phenomena are **topological**: they depend only on the winding number of the path, not microscopic details.
