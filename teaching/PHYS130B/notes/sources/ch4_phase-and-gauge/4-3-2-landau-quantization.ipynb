## Lecture Notes

Landau quantization describes the energy spectrum of a charged particle in a uniform magnetic field quantum mechanically. This is a cornerstone result: it shows that the continuous kinetic energy of a free particle becomes quantized into discrete *Landau levels*, each with enormous degeneracy. This degeneracy is the key to understanding the quantum Hall effect.

### Quantum Hamiltonian in a Magnetic Field

#### Minimal Coupling and Gauge Choice

The quantum Hamiltonian for a charged particle in an electromagnetic field is:

$$


\hat{H} = \frac{1}{2m}\left(\hat{\boldsymbol{p}} - \frac{q}{c}\boldsymbol{A}\right)^2 + q\Phi

$$


(In SI units, drop the $1/c$ factor.) For a uniform magnetic field $\boldsymbol{B} = B\hat{z}$ with no electric potential, $\Phi = 0$.

The energy depends on the choice of gauge for $\boldsymbol{A}$. Two convenient choices are:

:::{admonition} Gauge Choices
:class: note

1. **Symmetric (Landau) gauge**: $\boldsymbol{A} = \frac{B}{2}(-y, x, 0)$. This gauge respects rotational symmetry.
2. **Landau gauge**: $\boldsymbol{A} = (0, Bx, 0)$. This breaks translational symmetry in $x$ but simplifies algebra.

Both give the same physical spectrum; choice is a matter of convenience.
:::

#### Symmetric Gauge Approach

In the symmetric gauge, the kinetic momentum components are:

$$


\hat{\pi}_x = \hat{p}_x + \frac{qB}{2c}y, \quad \hat{\pi}_y = \hat{p}_y - \frac{qB}{2c}x

$$


The Hamiltonian becomes:

$$


\hat{H} = \frac{1}{2m}\left(\hat{\pi}_x^2 + \hat{\pi}_y^2\right) + \frac{\hat{p}_z^2}{2m}

$$


Motion along $z$ is unaffected. For motion in the $xy$-plane, define the cyclotron frequency:

$$


\omega_c = \frac{qB}{mc}

$$


The kinetic momentum components satisfy:

$$


[\hat{\pi}_x, \hat{\pi}_y] = -\mathrm{i}\hbar\frac{qB}{c} = -\mathrm{i}\hbar m\omega_c

$$


#### Ladder Operator Construction

The perpendicular kinetic momentum components look like position and momentum in harmonic oscillator form. Define ladder operators:

$$


\hat{a} = \sqrt{\frac{m\omega_c}{2\hbar}}\hat{\pi}_x + \mathrm{i}\sqrt{\frac{1}{2\hbar m\omega_c}}\hat{\pi}_y

$$


$$

\hat{a}^\dagger = \sqrt{\frac{m\omega_c}{2\hbar}}\hat{\pi}_x - \mathrm{i}\sqrt{\frac{1}{2\hbar m\omega_c}}\hat{\pi}_y

$$


These satisfy the standard commutation relation:

$$


[\hat{a}, \hat{a}^\dagger] = 1

$$


:::{admonition} Hamiltonian in Ladder Operator Form
:class: important

$$


\hat{H}_{\text{2D}} = \hbar\omega_c\left(\hat{a}^\dagger\hat{a} + \tfrac{1}{2}\right)

$$


This is exactly the Hamiltonian of a 2D harmonic oscillator, with each Landau level separated by energy $\hbar\omega_c$.
:::

### Landau Levels and Energy Spectrum

The eigenvalue problem is:

$$


\hat{H}|n\rangle = E_n|n\rangle

$$


with $\hat{a}^\dagger\hat{a}|n\rangle = n|n\rangle$ for $n = 0, 1, 2, \ldots$

:::{admonition} Landau Level Energies
:class: important

$$


E_n = \left(n + \tfrac{1}{2}\right)\hbar\omega_c

$$


All states with the same $n$ belong to the $n$-th *Landau level* (LL). There is no energy separation between states with different guiding center positions—only the quantum number $n$ matters.
:::

The ground state $n = 0$ has energy $E_0 = \frac{1}{2}\hbar\omega_c$ (the zero-point energy). Higher Landau levels are spaced by $\hbar\omega_c$ intervals.

### Magnetic Length and Degeneracy

The characteristic length scale in this problem is the **magnetic length**:

:::{admonition} Magnetic Length
:class: important

$$


\ell_B = \sqrt{\frac{\hbar}{qB}} = \sqrt{\frac{\hbar c}{qBc}}

$$


(In Gaussian units: $\ell_B = \sqrt{\hbar/eB}$.)

The magnetic length sets the size of the quantum wavefunction in the lowest Landau level. Two distinct guiding center positions separated by $\gtrsim \ell_B$ correspond to orthogonal wavefunctions.
:::

For a sample of area $A$, the degeneracy of each Landau level is:

:::{admonition} Landau Level Degeneracy
:class: important

$$


N_{\phi} = \frac{\Phi_{\text{total}}}{\Phi_0} = \frac{BA}{\Phi_0}

$$


where $\Phi_0 = \frac{hc}{q} = \frac{h}\mathrm{e}$ is the flux quantum (using $q = e$ for electrons).

Equivalently:

$$


N_{\phi} = \frac{A}{\pi \ell_B^2}

$$


Each Landau level can accommodate $N_{\phi}$ quantum states—a macroscopic degeneracy.
:::

:::{admonition} Derivation: Degeneracy from Uncertainty Principle
:class: dropdown information

In the lowest Landau level ($n = 0$), the guiding center position uncertainty is $\Delta X \sim \ell_B$. The number of distinct guiding center positions that fit in a region of area $A$ is approximately $N \sim A/(\pi\ell_B^2)$.

More rigorously, each state can be labeled by its guiding center position $(X, Y)$, and the separation between orthogonal states is $\ell_B$. Thus the degeneracy is the number of lattice points in an $\ell_B$-spacing lattice covering area $A$.
:::


:::{admonition} Discussion
:class: dropdown tip

The degeneracy of each Landau level, $N_\phi = BA/\Phi_0$, grows linearly with the magnetic field strength $B$. This is counterintuitive: stronger fields should confine particles more tightly, not create more states. What is really happening?

(a) The magnetic length $\ell_B = \sqrt{\hbar/(qB)}$ *decreases* as $B$ increases. How does the number of lattice sites (separated by $\ell_B$) that fit in a fixed area $A$ scale with decreasing $\ell_B$?

(b) A sample with area $A = 1 \text{ mm}^2 = 10^{-6} \text{ m}^2$ at $B = 1 \text{ T}$ has $N_\phi \approx 10^{10}$ degenerate states in the lowest Landau level. If you increase $B$ to $10 \text{ T}$ while keeping the sample size fixed, what happens to $N_\phi$? Estimate the new value.

(c) If all $N_\phi$ states have the same energy (to leading order), what prevents the ground state energy from being completely determined by the number of particles $N_e$? Why is the filling factor $\nu = N_e/N_\phi$ the crucial quantity?
:::

:::{admonition} Macroscopic Degeneracy and Its Significance
:class: note

- The degeneracy $N_{\phi} \sim 10^{10}$ for a typical Hall bar (area $\sim 1 \text{ mm}^2$ in a $B \sim 1 \text{ T}$ field).
- All $N_{\phi}$ states in a Landau level have the *same energy* (to leading order).
- This massive degeneracy is broken by disorder, interactions, and finite size effects.
- The degeneracy is the origin of the integer quantum Hall effect: it allows distinct edge states to carry independent current.
:::

#### Quantum vs Classical

Landau quantization is fundamentally quantum:
- Classically, all energies up to the cyclotron energy are allowed.
- Quantum mechanically, only discrete Landau levels appear, spaced by $\hbar\omega_c$.
- The gap between Landau levels grows with $B$; at very strong fields, only the lowest level is populated.

:::{admonition} Key Results: Landau Quantization
:class: important

- Perpendicular motion in a magnetic field is equivalent to a 2D harmonic oscillator.
- Energy levels: $E_n = (n + \tfrac{1}{2})\hbar\omega_c$ with $n = 0, 1, 2, \ldots$
- Each level is macroscopically degenerate: $N_{\phi} = BA/\Phi_0$ states per level.
- Magnetic length $\ell_B = \sqrt{\hbar/(qB)}$ characterizes wavefunction size.
- Landau quantization is the basis for quantum Hall physics and topological quantum phenomena.
:::
