## Lecture Notes

Gauge invariance is not merely a mathematical symmetry—it is the principle that guarantees all physical predictions are independent of unobservable choices. In this section, we explore how gauge freedom is exploited to simplify calculations, and which quantities are truly gauge-invariant (observable) versus gauge-dependent (unobservable).

### Gauge Choices and Computational Simplification

The freedom to choose a gauge is a powerful tool. Different problems are often easiest in different gauges.

#### Common Gauge Choices

**Coulomb Gauge**: $\nabla \cdot \boldsymbol{A} = 0$

Advantages:
- The scalar potential $\Phi$ is determined instantaneously by the charge density: $\nabla^2\Phi = -\rho/\epsilon_0$.
- Natural for non-relativistic QM with static charge distributions.
- Separates the longitudinal electric field (from charges) from the transverse field (from currents).

:::{admonition} Example: Coulomb Gauge in Atomic Physics
:class: example

In an atom, the Coulomb gauge makes the Coulomb potential appear explicitly and separately from the radiation field effects. This is ideal for calculating atomic structure with radiation as a perturbation.
:::

**Landau Gauge**: $A_x = 0, A_y = Bx$ (for a uniform field $B = B\hat{z}$)

Advantages:
- Explicitly breaks translational symmetry in one direction, simplifying separation of variables.
- Natural for calculating Landau levels (quantum Hall effect).
- Eigenfunctions have simpler form (plane waves in one direction).

**Symmetric (Landau) Gauge**: $\boldsymbol{A} = \frac{1}{2}\boldsymbol{B} \times \boldsymbol{r}$

Advantages:
- Manifests rotational symmetry about the magnetic field axis.
- Natural for problems with cylindrical symmetry.

#### Gauge Transformations Between Choices

Transforming from one gauge to another involves choosing a scalar function $\chi(\boldsymbol{r}, t)$ such that:

$$


\boldsymbol{A}' = \boldsymbol{A} - \nabla\chi, \quad \Phi' = \Phi + \frac{1}{q}\partial_t\chi

$$


The wavefunction transforms as $\psi' = \mathrm{e}^{\mathrm{i}q\chi/(\hbar c)}\psi$ to keep the physics unchanged.

:::{admonition} Gauge Transformations and Simplification
:class: note

Although different gauges simplify different calculations, the energy eigenvalues, transition rates, and all observable predictions must be identical. If they are not, an error has been made—gauge invariance provides a consistency check.
:::

### Gauge-Invariant and Gauge-Dependent Quantities

Not all quantum mechanical quantities are observable. Some depend on the gauge choice.

:::{admonition} Classification of Observables
:class: important

**Gauge-Invariant (Observable):**
- Probability density: $|\psi|^2$
- Probability current: $\boldsymbol{j} = \text{Im}(\psi^*\hat{\boldsymbol{\pi}}\psi)/m$
- Energy eigenvalues: $E_n$
- Transition probabilities: $|\langle n|O|m \rangle|^2$
- Magnetic flux through a loop: $\oint \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S}$
- **Wilson loop**: $W(C) = \exp(\mathrm{i}q\oint_C \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}/(\hbar c))$ (for certain geometric configurations)

**Gauge-Dependent (Unobservable in isolation):**
- Vector potential: $\boldsymbol{A}(\boldsymbol{r})$
- Scalar potential: $\Phi(\boldsymbol{r})$
- Phase of the wavefunction: $\arg[\psi(\boldsymbol{r})]$ at a single point
- Canonical momentum: $\hat{\boldsymbol{p}}$
- Individual Aharonov-Bohm phase along a single path
:::

The key insight: quantities that are intrinsically gauge-dependent can still combine in gauge-invariant ways. For example, the relative phase difference between two paths is gauge-invariant even though individual path phases are gauge-dependent.


:::{admonition} Discussion
:class: dropdown tip

Gauge transformations leave measurable quantities invariant, yet different gauges give different vector potentials and scalar potentials. Suppose you solve the Landau level problem (cyclotron motion in a magnetic field) in two different gauges: the symmetric gauge $\boldsymbol{A} = \frac{1}{2}\boldsymbol{B} \times \boldsymbol{r}$ and the Landau gauge $\boldsymbol{A} = (0, Bx, 0)$.

(b) Would the wavefunctions $\psi_{n}(\boldsymbol{r})$ have the same functional form in both gauges, or would they look completely different? If different, why must physical predictions (transition rates, probability density integrals) still agree?

(c) How would you verify explicitly that a gauge transformation from one to the other does not change an observable like the probability current $\boldsymbol{j} = \text{Im}(\psi^*\hat{\boldsymbol{\pi}}\psi)/m$?
:::

### Wilson Loops and Non-Abelian Gauge Theory

A **Wilson loop** is a path-ordered integral of the gauge field around a closed contour:

$$


W(C) = \mathcal{P}\exp\left[\frac{\mathrm{i}q}{\hbar c}\oint_C \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}\right]

$$


For an abelian gauge (electromagnetism), this simplifies to:

$$


W(C) = \exp\left[\frac{\mathrm{i}q}{\hbar c}\oint_C \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}\right] = \exp\left[\frac{\mathrm{i}q\Phi_B}{\hbar c}\right]

$$


where $\Phi_B$ is the flux enclosed by the loop.

:::{admonition} Wilson Loop as Gauge-Invariant Observable
:class: important

The Wilson loop $W(C)$ is **gauge-invariant**: it depends only on the flux enclosed by $C$, not on the detailed form of $\boldsymbol{A}$.

In non-abelian gauge theories (like QCD), Wilson loops play a central role in understanding confinement and the quark-gluon dynamics of the strong nuclear force.
:::

### Peierls Substitution: Gauge Fields on a Lattice

For a tight-binding model on a lattice, the Peierls substitution incorporates a gauge field by modifying the hopping amplitudes:

$$


c_j^\dagger c_{j+1} \to \exp\left[\frac{\mathrm{i}q}{\hbar c}\int_j^{j+1}\boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}\right] c_j^\dagger c_{j+1}

$$


This gauge-invariant coupling naturally incorporates magnetic effects into condensed matter models.

:::{admonition} Example: Lattice Electrons in a Magnetic Field
:class: example

In a 2D square lattice with nearest-neighbor hopping $t$, applying a perpendicular magnetic field creates a spatial modulation of phases. The Peierls substitution correctly captures this, leading to fractal energy spectra (Hofstadter butterfly) and eventually to integer and fractional quantum Hall effects at strong fields.
:::

### Universal Properties: Independence of Microscopic Details

Some quantities are so fundamental that they depend only on topological (global) properties, not microscopic details. This topological robustness is a hallmark of gauge-invariant physics.

:::{admonition} Topological Protection of Observables
:class: dropdown

Consider the integer quantum Hall conductance $\sigma = ne^2/h$ for $n$ filled Landau levels. This is:
- **Independent of disorder** (within each Landau level; disorder causes broadening but not reduction of plateaus)
- **Independent of microscopic details** (material composition, electron density variations, sample shape)
- **Quantized to integer multiples** of $e^2/h$

Why? Because the Hall conductance is fundamentally a topological invariant—the integrated Berry curvature over the Brillouin zone. Gauge invariance ensures that only topological properties affect universal quantities.

This is why quantum Hall plateaus are so robust and have become standards for electrical resistance calibration worldwide.
:::

### Gauge Invariance of the Schrödinger Equation Revisited

Finally, let's verify explicitly that the Schrödinger equation itself is gauge-covariant.

:::{admonition} Derivation: Covariance of the Schrödinger Equation
:class: dropdown information

Suppose $\psi(\boldsymbol{r}, t)$ satisfies:

$$


\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi\right]\psi

$$


Under gauge transformation $\psi \to \psi' = \mathrm{e}^{\mathrm{i}q\chi/(\hbar c)}\psi$ with $\boldsymbol{A} \to \boldsymbol{A}' = \boldsymbol{A} - \nabla\chi$ and $\Phi \to \Phi' = \Phi + \partial_t\chi/q$:

The time derivative becomes:

$$


\frac{\partial\psi'}{\partial t} = \mathrm{i}\frac{q}{\hbar c}\frac{\partial\chi}{\partial t}\mathrm{e}^{\mathrm{i}q\chi/(\hbar c)}\psi + \mathrm{e}^{\mathrm{i}q\chi/(\hbar c)}\frac{\partial\psi}{\partial t}

$$


The spatial derivative involves:

$$


(\hat{\boldsymbol{p}} - q\boldsymbol{A}')\psi' = \left[\hat{\boldsymbol{p}} - q(\boldsymbol{A} - \nabla\chi)\right]\mathrm{e}^{\mathrm{i}q\chi/(\hbar c)}\psi

$$


After careful calculation (collecting terms from the product rule), the $\chi$ terms cancel perfectly, and $\psi'$ satisfies the equation with the primed potentials.

The Schrödinger equation has the **same form** in all gauges—it is manifestly gauge-covariant.
:::

### Summary

- Different **gauge choices** simplify different problems: Coulomb gauge for atoms, Landau gauge for Landau levels, symmetric gauge for cylindrical symmetry.
- **Gauge-invariant** quantities (energy, transition rates, flux) are observable; **gauge-dependent** quantities ($\boldsymbol{A}$, $\Phi$, individual phases) are not.
- **Wilson loops** and **topological invariants** encode gauge-invariant information about enclosed flux and field configurations.
- **Peierls substitution** incorporates gauge fields into lattice models in a gauge-invariant way.
- **Topological robustness** explains why integer quantum Hall plateaus and other universal features are independent of microscopic disorder and details.
- The Schrödinger equation is **gauge-covariant**: it has identical form in all gauges, confirming that physics is gauge-invariant.
