## Lecture Notes

### Overview

The quantum Hall effect reveals one of the most robust and universal phenomena in condensed matter physics. When a 2D electron gas is placed in a strong perpendicular magnetic field, the Hall conductance becomes quantized in units of $e^2/h$. Integer quantum Hall effect (IQHE) occurs when Landau levels are completely filled; fractional quantum Hall effect (FQHE) occurs at rational filling factors, where interactions dominate. The charge pumping argument elegantly shows why quantization emerges and connects gauge physics to topological properties.

### Filling Landau Levels

In a perpendicular magnetic field $B$, electron energy levels organize into Landau levels, each with degeneracy $N_\phi = BA\Phi_0^{-1} = BAe/h$, where $A$ is the sample area.

:::{admonition} Filling Factor
:class: important

The filling factor is

$$
\nu = \frac{N_e}{N_\phi}
$$

where $N_e$ is the number of electrons and $N_\phi$ is the number of flux quanta. When $\nu$ is an integer, exactly $\nu$ Landau levels are completely filled, and the Fermi level lies in a gap. The energy cost to excite an electron across the gap is $\hbar\omega_c$ (cyclotron energy), which is crucial for protecting quantization against disorder.
:::

### Charge Pumping Argument

The most elegant derivation uses Laughlin's **charge pumping** picture. Consider a Corbino disk: a 2D sample shaped like an annulus, with inner and outer edges. Imagine slowly threading a single flux quantum $\Phi_0 = h/e$ through the hole, adiabatically, while maintaining filled Landau levels. As the flux increases, $N_\phi$ increases by exactly one.

**Key insight:** If $N_e$ is fixed and $N_\phi$ increases by 1, the system must adjust. Since one Landau level is completely filled and a new flux quantum appears, one electron must be "pumped" out to make room. This pumping occurs adiabatically at the edge.

:::{admonition} Charge Pumping and Quantized Conductance
:class: important

When one flux quantum $\Phi_0$ is threaded through the sample at filling $\nu$, exactly $\nu$ electrons are pumped out. This pumping is adiabatic: the Landau level gap remains open, preventing level mixing.

For each pumped charge $\nu e$, the Hall conductance is:

$$
\sigma_{xy} = \nu \frac{e^2}{h}
$$

This is exact and depends only on $\nu$, $e$, $h$—not on disorder, sample geometry, or microscopic details. The quantization emerges from a global constraint (flux conservation and filling factor) combined with a gap that prevents local scattering from destroying the state.
:::

### Linear Response and Hall Conductance

Apply a weak oscillatory electric field $\boldsymbol{E}(t) = \boldsymbol{E}\,\mathrm{e}^{-\mathrm{i}(\omega + \mathrm{i}0^+)t}$ adiabatically switched on from the infinite past. First-order time-dependent perturbation theory yields the **Kubo formula** for the Hall conductivity, which gives the same quantization $\sigma_{xy} = \nu\, e^2/h$ from a completely different approach.

:::{admonition} Kubo Formula for Hall Conductivity
:class: important

$$
\sigma_H = \frac{1}{\mathrm{i} N_\phi} \sum_{\substack{n,m \in \text{occ}}} \sum_{n' \neq n} \frac{\langle n,m \vert \hat{\pi}_x \vert n',m \rangle \langle n',m \vert \hat{\pi}_y \vert n,m \rangle - \text{h.c.}}{(E_{n'} - E_n)^2}
$$ (eq-kubo-hall)

where $\vert n,m\rangle$ are Landau level states, $E_n = \hbar\omega_c(n + \tfrac{1}{2})$, and $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - e\boldsymbol{A}/c$ is the kinetic momentum.

Evaluating this sum gives $\sigma_H = \nu \frac{e^2}{h}$: each occupied state contributes exactly $1/N_\phi$ to the Hall conductivity.
:::

:::{admonition} Derivation: Kubo Formula Evaluation
:class: dropdown information

In units $m = e = \hbar = 1$, the Landau level Hamiltonian is $\hat{H} = \tfrac{1}{2}\hat{\boldsymbol{\pi}}^2$ with energies $E_n = B(n + \tfrac{1}{2})$.

**Setup:** An oscillatory electric field $\boldsymbol{E}(t) = \boldsymbol{E}\,\mathrm{e}^{-\mathrm{i}(\omega + \mathrm{i}0^+)t}$ corresponds to a vector potential perturbation

$$
\delta\boldsymbol{A}(t) = -\frac{\boldsymbol{E}}{\mathrm{i}\omega}\,\mathrm{e}^{-\mathrm{i}(\omega + \mathrm{i}0^+)t}
$$

which perturbs $\hat{H}$ as $\delta\hat{H}(t) = -\hat{\boldsymbol{\pi}} \cdot \delta\boldsymbol{A}(t)$.

**First-order perturbation:** Time-dependent perturbation theory (Dyson series) gives the current density $\langle \boldsymbol{j}(t)\rangle = \langle \boldsymbol{j}^{(0)}\rangle + \langle \boldsymbol{j}^{(1)}(t)\rangle$. The zeroth-order term vanishes because $\langle n,m\vert\hat{\boldsymbol{\pi}}\vert n,m\rangle = 0$ ($\hat{\pi}_x, \hat{\pi}_y$ are off-diagonal in Landau level index). The first-order term yields:

$$
\sigma_{ji}(\omega) = \frac{\mathrm{i}B}{\omega N_\phi} \sum_{n,m \in \text{occ}} \sum_{n' \neq n} \left(\frac{\langle n,m\vert \hat{\pi}_j\vert n',m\rangle\langle n',m\vert \hat{\pi}_i\vert n,m\rangle}{\omega - E_n + E_{n'}} - \frac{\langle n,m\vert \hat{\pi}_i\vert n',m\rangle\langle n',m\vert \hat{\pi}_j\vert n,m\rangle}{\omega + E_n - E_{n'}}\right)
$$

**DC limit:** Expand each denominator to first order in $\omega$. The antisymmetric piece (contributing to Hall conductivity) is:

$$
\sigma_H = \frac{B}{\mathrm{i}N_\phi} \sum_{n,m \in \text{occ}} \sum_{n' \neq n} \frac{\langle n,m\vert \hat{\pi}_x\vert n',m\rangle\langle n',m\vert \hat{\pi}_y\vert n,m\rangle - \text{h.c.}}{(E_{n'} - E_n)^2}
$$

**Ladder operators:** Using $\hat{\pi}_x = \sqrt{B/2}\,(\hat{a}^\dagger + \hat{a})$ and $\hat{\pi}_y = \mathrm{i}\sqrt{B/2}\,(\hat{a}^\dagger - \hat{a})$:

$$
\langle n,m\vert \hat{\pi}_x\vert n',m\rangle\langle n',m\vert \hat{\pi}_y\vert n,m\rangle = \frac{\mathrm{i}B}{2}\Big((n+1)\,\delta_{n', n+1} - n\,\delta_{n', n-1}\Big)
$$

Since $E_{n'} - E_n = B(n' - n)$, both denominators equal $B^2$, so:

$$
\sigma_H = \frac{1}{N_\phi} \sum_{n,m \in \text{occ}} \Big((n+1) - n\Big) = \frac{1}{N_\phi} \sum_{n,m \in \text{occ}} 1
$$

For $\nu$ completely filled Landau levels: $\sigma_H = \nu\, e^2/h$. $\checkmark$
:::

:::{admonition} Discussion
:class: dropdown tip

Why do the charge pumping argument and the Kubo formula agree, despite using completely different reasoning?
- Charge pumping is **topological**: it counts electron transfer when flux is threaded, independent of the microscopic Hamiltonian.
- Kubo is **dynamical**: it sums current operator matrix elements between all occupied and unoccupied states.

What property of Landau levels guarantees that each level contributes exactly $e^2/h$?
:::

### Fractional Quantum Hall Effect

At fractional fillings $\nu = p/q$ (coprime integers), many-body electron-electron interactions become essential. The ground state is a highly entangled quantum fluid with no simple mean-field description.

:::{admonition} Laughlin Wavefunction
:class: note

For $\nu = 1/m$ (odd $m$), Laughlin proposed:

$$
\Psi_{\text{Laughlin}} = \prod_{i<j} (z_i - z_j)^m \exp\left(-\frac{1}{4\ell_B^2}\sum_i |z_i|^2\right)
$$

The factor $(z_i - z_j)^m$ enforces strong repulsion between electrons, creating an incompressible quantum fluid with a gap to excitations.
:::

The charge pumping argument extends: when one flux quantum is added at $\nu = p/q$, exactly $p/q$ of an electron is pumped out. This fractional pumping is possible because the quantum fluid redistributes charge non-locally. The **quasiparticles** (hole excitations) carry fractional charge $e/q$ and obey **fractional statistics**—the anyons from [2.3.2 Fractional Quantum Hall Anyons](../ch2_identical-particles/2-3-2-fractional-quantum-hall-anyons).

:::{admonition} Hall Conductance at Fractional Filling
:class: important

At fractional filling $\nu = p/q$, the Hall conductance is:

$$
\sigma_{xy} = \nu \frac{e^2}{h} = \frac{p}{q} \frac{e^2}{h}
$$

Quasiparticles carry charge $\pm e/q$ and obey fractional Bragg statistics: exchanging two quasiparticles introduces a phase $\exp(2\pi \mathrm{i} p/q)$.
:::


:::{admonition} Poll: Hall resistance quantization
:class: dropdown poll

In the quantum Hall effect, the Hall resistance is quantized as $R_H = h / (q^2 \nu)$ where $\nu$ is the filling factor (ratio of electrons to flux quanta). The remarkable feature is that $R_H$ depends only on fundamental constants and the filling factor, independent of material details like the sample geometry or the magnetic field strength (over wide ranges). Which statement correctly explains this universal quantization?

(A) At integer filling factors, all Landau levels below a certain energy are completely filled, so the system is insulating and all electrons contribute equally to transport.

(B) The quantized nature of Landau levels means only a fixed number of electrons can carry current, regardless of other parameters.

(C) The Hall resistance is determined by topological properties of the electron states, which are robust against disorder and material variations.

(D) Both (A) and (C).

:::

### Summary

- **Filling factor** $\nu = N_e/N_\phi$ counts filled Landau levels; integer $\nu$ means a gap protects the state.
- **Charge pumping:** Threading a flux quantum through an adiabatic cycle pumps out $\nu$ electrons, giving exact quantization $\sigma_{xy} = \nu e^2/h$.
- **Robustness** comes from the Landau level gap: disorder cannot mix states within a filled level, and quantization depends only on global filling.
- **Fractional QHE** at $\nu = p/q$ produces incompressible quantum fluids with fractional charge and fractional statistics.
- The QHE unifies **gauge physics**, **topology**, and **many-body physics** (Laughlin wavefunction and anyons).
