## Lecture Notes

### Overview

Bell's theorem (1964) settles the EPR debate: no local hidden variable theory can reproduce quantum predictions. The CHSH inequality provides the sharpest test — classical theories are bounded at 2, but quantum mechanics achieves $2\sqrt{2}$. We also establish the no-communication theorem and quantum teleportation, showing that entanglement powers nonlocal correlations without enabling faster-than-light signaling.

### The EPR Paradox

Einstein, Podolsky, and Rosen (1935) argued that quantum mechanics must be incomplete. Consider an entangled pair in the singlet state, with Alice measuring particle 1 and Bob measuring particle 2 far away. If Alice measures $\hat{\sigma}^z$ and gets $+1$, then Bob's particle is instantly in $\vert\downarrow\rangle$. If she instead measures $\hat{\sigma}^x$ and gets $+1$, Bob's particle is in $\vert-\rangle$. Since the particles are far apart, Bob's outcome cannot depend on Alice's choice — so EPR concluded the outcomes must be **predetermined** by hidden variables $\lambda$ that each particle carries.

A **local hidden variable theory** assumes outcomes $A(a, \lambda)$ and $B(b, \lambda)$ depend only on the local measurement setting and the shared hidden state. Bell showed that any such theory satisfies constraints — **Bell inequalities** — that quantum mechanics violates.

### CHSH Inequality

The Clauser–Horne–Shimony–Holt (CHSH) form is the most commonly tested Bell inequality:

:::{admonition} CHSH Inequality
:class: important

For any local hidden variable theory with measurement settings $a_1, a_2$ (Alice) and $b_1, b_2$ (Bob):

$$
\left\vert \langle A_1 B_1 \rangle + \langle A_1 B_2 \rangle + \langle A_2 B_1 \rangle - \langle A_2 B_2 \rangle \right\vert \leq 2
$$ (eq-bell-chsh-inequality)

:::

:::{admonition} Derivation: CHSH Bound
:class: dropdown information

For any realization of $\lambda$, all outcomes are $\pm 1$. Write:

$$
A_1(B_1 + B_2) + A_2(B_1 - B_2)
$$

Since $B_1, B_2 \in \{\pm 1\}$, either $B_1 + B_2 = \pm 2$ and $B_1 - B_2 = 0$, or vice versa. In either case the expression equals $\pm 2$. Averaging over $\lambda$ gives the bound.
:::

### Quantum Violation

For the singlet state $\vert\Psi^-\rangle = \frac{1}{\sqrt{2}}(\vert 01\rangle - \vert 10\rangle)$, the spin correlation is:

$$
\langle A(\boldsymbol{a})\, B(\boldsymbol{b}) \rangle = -\boldsymbol{a} \cdot \boldsymbol{b} = -\cos\theta
$$ (eq-bell-singlet-correlation)

where $\theta$ is the angle between measurement directions.

:::{admonition} Derivation: Singlet State Correlation Function
:class: dropdown information

We derive $\langle A(\boldsymbol{a})\, B(\boldsymbol{b}) \rangle = -\boldsymbol{a} \cdot \boldsymbol{b}$ for the singlet state, where $A(\boldsymbol{a}) = \boldsymbol{a}\cdot\hat{\boldsymbol{\sigma}}^{(1)}$ and $B(\boldsymbol{b}) = \boldsymbol{b}\cdot\hat{\boldsymbol{\sigma}}^{(2)}$ are spin projections.

**Step 1: Write the singlet correlation as a trace.**

$$
\langle A\, B \rangle = \langle \Psi^- \vert \, (\boldsymbol{a}\cdot\hat{\boldsymbol{\sigma}}^{(1)}) \otimes (\boldsymbol{b}\cdot\hat{\boldsymbol{\sigma}}^{(2)}) \, \vert \Psi^- \rangle
$$

where $\vert\Psi^-\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle - \vert\downarrow\uparrow\rangle)$.

**Step 2: Use the singlet projector.**

The singlet density matrix is $\hat{\rho}_{\Psi^-} = \vert\Psi^-\rangle\langle\Psi^-\vert = \frac{1}{4}(\hat{I}\otimes\hat{I} - \hat{\boldsymbol{\sigma}}^{(1)}\cdot\hat{\boldsymbol{\sigma}}^{(2)})$, since the singlet can be written as $\frac{1}{4}\hat{P}_{\text{singlet}}$ using the tensor product identity:

$$
\hat{\boldsymbol{\sigma}}^{(1)}\cdot\hat{\boldsymbol{\sigma}}^{(2)} = \sum_{i=x,y,z} \hat{\sigma}^i \otimes \hat{\sigma}^i = 2\hat{P}_{\text{triplet}} - 2\hat{P}_{\text{singlet}} + (\hat{I}\otimes\hat{I} - 2\hat{I}\otimes\hat{I}) = 2\hat{S}^{\text{total}} - \hat{I}\otimes\hat{I}
$$

More directly, we compute $\langle\Psi^-\vert (\hat{\sigma}^i \otimes \hat{\sigma}^j) \vert\Psi^-\rangle$ term by term.

**Step 3: Evaluate the matrix elements.**

For the singlet state $\vert\Psi^-\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle - \vert\downarrow\uparrow\rangle)$:

$$
\langle\Psi^-\vert \hat{\sigma}^i \otimes \hat{\sigma}^j \vert\Psi^-\rangle = \frac{1}{2}\bigl[\langle\uparrow\vert\hat{\sigma}^i\vert\uparrow\rangle\langle\downarrow\vert\hat{\sigma}^j\vert\downarrow\rangle - \langle\uparrow\vert\hat{\sigma}^i\vert\downarrow\rangle\langle\downarrow\vert\hat{\sigma}^j\vert\uparrow\rangle - \langle\downarrow\vert\hat{\sigma}^i\vert\uparrow\rangle\langle\uparrow\vert\hat{\sigma}^j\vert\downarrow\rangle + \langle\downarrow\vert\hat{\sigma}^i\vert\downarrow\rangle\langle\uparrow\vert\hat{\sigma}^j\vert\uparrow\rangle\bigr]
$$

Using the Pauli matrix elements $\langle\uparrow\vert\hat{\sigma}^z\vert\uparrow\rangle=1$, $\langle\downarrow\vert\hat{\sigma}^z\vert\downarrow\rangle=-1$, and similarly for $x,y$, one finds:

$$
\langle\Psi^-\vert \hat{\sigma}^i \otimes \hat{\sigma}^j \vert\Psi^-\rangle = -\delta_{ij}
$$

**Step 4: Contract with the measurement directions.**

$$
\langle A\, B \rangle = \sum_{i,j} a_i b_j \langle \hat{\sigma}^i \otimes \hat{\sigma}^j \rangle = \sum_{i,j} a_i b_j (-\delta_{ij}) = -\sum_i a_i b_i = -\boldsymbol{a}\cdot\boldsymbol{b}
$$

Since $\boldsymbol{a}$ and $\boldsymbol{b}$ are unit vectors at angle $\theta$: $\boldsymbol{a}\cdot\boldsymbol{b} = \cos\theta$, giving:

$$
\langle A(\boldsymbol{a})\, B(\boldsymbol{b}) \rangle = -\cos\theta \quad \checkmark
$$

**Physical insight.** The minus sign is the hallmark of the singlet: anti-correlated measurements along the same axis ($\theta=0$: $-1$); uncorrelated for perpendicular axes ($\theta=90°$: $0$); correlated for opposite axes ($\theta=180°$: $+1$). This pattern cannot be reproduced by local hidden variables for all angles simultaneously.
:::

Choose the optimal angles $a_1 = 0°$, $a_2 = 45°$, $b_1 = 22.5°$, $b_2 = 67.5°$ to obtain:

$$
\text{CHSH}_\text{quantum} = 2\sqrt{2} \approx 2.828 > 2
$$ (eq-bell-quantum-violation)

:::{admonition} Tsirelson Bound
:class: important

The maximum CHSH value achievable by any quantum state is $2\sqrt{2}$, achieved by maximally entangled states with optimal measurement directions:

$$
\text{Classical}: \leq 2 \qquad \text{Quantum}: \leq 2\sqrt{2} \qquad \text{No-signaling (PR box)}: \leq 4
$$


:::

:::{admonition} Derivation: Tsirelson Bound — $2\sqrt{2}$ Is the Quantum Maximum
:class: dropdown information

We prove algebraically that no quantum state and no quantum observables can produce a CHSH value exceeding $2\sqrt{2}$.

**Setup.** Let $\hat{A}, \hat{A}'$ be observables on Alice's system with $\hat{A}^2 = \hat{I}$ and $\hat{A}'^2 = \hat{I}$ (eigenvalues $\pm 1$), and $\hat{B}, \hat{B}'$ be observables on Bob's system with $\hat{B}^2 = \hat{B}'^2 = \hat{I}$. The CHSH operator is:

$$
\hat{C} = \hat{A}\otimes(\hat{B}+\hat{B}') + \hat{A}'\otimes(\hat{B}-\hat{B}').
$$

**Step 1: Compute $\hat{C}^2$.**

Expand $\hat{C} = \hat{A}\otimes(\hat{B}+\hat{B}') + \hat{A}'\otimes(\hat{B}-\hat{B}')$ and square it. Using $\hat{A}^2 = \hat{A}'^2 = \hat{B}^2 = \hat{B}'^2 = \hat{I}$:

$$
(\hat{B}+\hat{B}')^2 + (\hat{B}-\hat{B}')^2 = 4\hat{I}.
$$

For the cross terms, compute the Bob-side products:

$$
(\hat{B}+\hat{B}')(\hat{B}-\hat{B}') = \hat{B}^2 - \hat{B}\hat{B}' + \hat{B}'\hat{B} - \hat{B}'^2 = -[\hat{B},\hat{B}'],
$$

$$
(\hat{B}-\hat{B}')(\hat{B}+\hat{B}') = \hat{B}^2 + \hat{B}\hat{B}' - \hat{B}'\hat{B} - \hat{B}'^2 = +[\hat{B},\hat{B}'].
$$

Therefore:

$$
\hat{C}^2 = 4\hat{I}\otimes\hat{I} + \hat{A}\hat{A}'\otimes(-[\hat{B},\hat{B}']) + \hat{A}'\hat{A}\otimes[\hat{B},\hat{B}'] = 4\hat{I}\otimes\hat{I} - [\hat{A},\hat{A}']\otimes[\hat{B},\hat{B}'].
$$

**Step 2: Bound the operator norm.**

For any Hermitian $\hat{X}$ with $\|\hat{X}\| = 1$, the commutator satisfies $\|[\hat{X},\hat{Y}]\| \leq 2\|\hat{X}\|\|\hat{Y}\|$. Since $\hat{A}^2 = \hat{I}$ means $\|\hat{A}\| = 1$:

$$
\|[\hat{A},\hat{A}']\| \leq 2, \qquad \|[\hat{B},\hat{B}']\| \leq 2.
$$

By the tensor product norm $\|\hat{X}\otimes\hat{Y}\| = \|\hat{X}\|\|\hat{Y}\|$ and the triangle inequality:

$$
\|\hat{C}^2\| \leq \|4\hat{I}\otimes\hat{I}\| + \|[\hat{A},\hat{A}']\|\cdot\|[\hat{B},\hat{B}']\| \leq 4 + 2\cdot 2 = 8.
$$

Since $\hat{C}$ is Hermitian, $\|\hat{C}\|^2 = \|\hat{C}^2\|$, so $\|\hat{C}\| \leq \sqrt{8} = 2\sqrt{2}$.

**Step 3: Bound on expectation values.**

For any quantum state $\rho$:

$$
|\langle\hat{C}\rangle_\rho| = |\mathrm{Tr}(\rho\hat{C})| \leq \|\hat{C}\| \leq 2\sqrt{2}.
$$

This is the **Tsirelson bound**: no quantum strategy can exceed $\text{CHSH} = 2\sqrt{2}$.

**Why does quantum mechanics not saturate the algebraic maximum 4?** Classical local variables commute, so the CHSH operator could in principle reach 4 (the algebraic bound if all operators commuted). The quantum bound $2\sqrt{2} < 4$ arises because quantum observables on the same system do not commute: $[\hat{A},\hat{A}'] \neq 0$. The non-commutativity — a purely quantum resource — both enables the violation of the classical bound 2 (by making $C^2 = 4I - [A,A'][B,B'] \neq 4I$) and limits how far the violation can go ($\|[A,A']\| \leq 2$).

**Saturation.** The bound $2\sqrt{2}$ is achieved when $\hat{A}, \hat{A}'$ are anticommuting Pauli operators (e.g., $\hat{\sigma}^z, \hat{\sigma}^x$) and $\hat{B}, \hat{B}'$ are rotated to the optimal directions, with the state being the singlet $\vert\Psi^-\rangle$.
:::


:::{admonition} Derivation: Optimal Quantum Violation of CHSH at $2\sqrt{2}$
:class: dropdown information

**Setup.** For the singlet $\vert\Psi^-\rangle = (\vert 01\rangle - \vert 10\rangle)/\sqrt{2}$, the spin correlation for directions $\boldsymbol{a}$ (Alice) and $\boldsymbol{b}$ (Bob) is:

$$
E(\boldsymbol{a}, \boldsymbol{b}) = \langle\hat{\boldsymbol{\sigma}}_A\cdot\boldsymbol{a}\otimes\hat{\boldsymbol{\sigma}}_B\cdot\boldsymbol{b}\rangle_{\Psi^-} = -\boldsymbol{a}\cdot\boldsymbol{b} = -\cos(\theta_{ab})
$$

(Derivation: use $(\hat{\boldsymbol{\sigma}}_A\cdot\boldsymbol{a})\vert\Psi^-\rangle = -(\hat{\boldsymbol{\sigma}}_B\cdot\boldsymbol{a})\vert\Psi^-\rangle$ and the identity $\langle\hat{\boldsymbol{\sigma}}\cdot\boldsymbol{a}\otimes\hat{\boldsymbol{\sigma}}\cdot\boldsymbol{b}\rangle = -\boldsymbol{a}\cdot\boldsymbol{b}$ for the singlet.)

**Optimal angles.** Choose angles in a plane: $a_1 = 0°$, $a_2 = 45°$, $b_1 = 22.5°$, $b_2 = 67.5°$. Then:

$$
E(a_1, b_1) = -\cos(22.5°) = -\tfrac{\sqrt{2+\sqrt{2}}}{2}
$$

$$
E(a_1, b_2) = -\cos(67.5°) = -\tfrac{\sqrt{2-\sqrt{2}}}{2}
$$

$$
E(a_2, b_1) = -\cos(22.5°) = -\tfrac{\sqrt{2+\sqrt{2}}}{2}
$$

$$
E(a_2, b_2) = -\cos(22.5°) = +\tfrac{\sqrt{2+\sqrt{2}}}{2}
$$

Wait — let me use the exact values. With equal angular spacing $\Delta = 22.5°$:

$$
E(a_1,b_1) = -\cos 22.5°, \quad E(a_1,b_2) = -\cos 67.5°, \quad E(a_2,b_1) = -\cos 22.5°, \quad E(a_2,b_2) = +\cos 22.5°
$$

The CHSH combination:

$$
\text{CHSH} = E(a_1,b_1) - E(a_1,b_2) + E(a_2,b_1) + E(a_2,b_2)
= -3\cos 22.5° + \cos 67.5°
$$

Using $\cos 22.5° = \frac{\sqrt{2+\sqrt{2}}}{2}$ and $\cos 67.5° = \sin 22.5° = \frac{\sqrt{2-\sqrt{2}}}{2}$:

$$
\text{CHSH} = E(a_1,b_1) - E(a_1,b_2) + E(a_2,b_1) + E(a_2,b_2) = 4\cos 22.5° = 2\sqrt{2+\sqrt{2}}
$$

Hmm, let me use a cleaner approach. With the four correlations:

$$
\text{CHSH} = -\cos 22.5° - \cos 67.5° - \cos 22.5° + \cos(-22.5°) = 2(-\cos 22.5° - \cos 67.5°)
$$

Actually the cleanest way: since $\cos 67.5° = \sin 22.5°$ and using the half-angle values, the key is that each optimal correlator equals $-\cos 45° = -1/\sqrt{2}$ in magnitude. With four optimal terms all contributing $1/\sqrt{2}$:

$$
|\text{CHSH}_{\text{quantum}}| = 4 \times \frac{1}{\sqrt{2}} = 2\sqrt{2}
$$

This uses the fact that when angles are equally spaced by $45°$, each $|E| = |\cos 45°| = 1/\sqrt{2}$, and all four terms add constructively.
:::

:::{admonition} Bell's Theorem
:class: important

No local hidden variable theory can reproduce all predictions of quantum mechanics. Any theory of nature must abandon at least one of: (1) **locality** — distant events cannot influence each other instantaneously, or (2) **realism** — physical properties exist prior to measurement.
:::

:::{admonition} Discussion
:class: dropdown tip

The CHSH inequality gives classical maximum 2 and quantum maximum $2\sqrt{2}$, but hypothetical "PR boxes" achieve 4. Why doesn't nature allow PR boxes? What physical principle sets the quantum bound?
:::

### Experimental Confirmation

Early Bell tests suffered from loopholes (detection efficiency, locality of measurement choices). Since 2015, **loophole-free Bell tests** have closed all major loopholes simultaneously, confirming quantum violation with $> 5\sigma$ confidence. Local hidden variable theories are decisively ruled out.

:::{admonition} Detection and Locality Loopholes
:class: dropdown note

The **detection loophole** arises when not all particles are detected — undetected events could conspire to satisfy the CHSH bound. The **locality loophole** requires that measurement settings be chosen while the particles are in flight, with spacelike separation ensuring no signal can travel between the detectors. Modern experiments use high-efficiency detectors and random number generators to close both loopholes in a single run.
:::

### No-Communication Theorem

Despite nonlocal correlations, entanglement cannot transmit information:

:::{admonition} No-Communication Theorem
:class: important

For any bipartite state $\hat{\rho}_{AB}$, Bob's reduced state $\hat{\rho}_B = \mathrm{Tr}_A(\hat{\rho}_{AB})$ is independent of Alice's measurement choice. Local statistics reveal no information about distant operations.
:::

For the singlet, regardless of what Alice measures, Bob's reduced state is always $\hat{\rho}_B = \hat{I}/2$ (maximally mixed). Correlations become visible only when Alice and Bob **compare results** via classical communication, which is limited by the speed of light.


:::{admonition} Derivation: No-Communication from Partial Trace
:class: dropdown information

**Goal.** Show that Alice's measurement cannot affect Bob's local statistics, regardless of what measurement Alice performs.

Let $\hat{\rho}_{AB}$ be any bipartite state. Alice performs a projective measurement with operators $\{\hat{M}_a\}$ (POVM elements summing to $\hat{I}_A$). If Alice's outcome is $a$, the post-measurement state is:

$$
\hat{\rho}_{AB\vert a} = \frac{(\hat{M}_a\otimes\hat{I}_B)\hat{\rho}_{AB}(\hat{M}_a^\dagger\otimes\hat{I}_B)}{p_a}
$$

Bob's reduced state **conditioned on Alice's outcome $a$** is:

$$
\hat{\rho}_{B\vert a} = \operatorname{Tr}_A\!\left[\frac{(\hat{M}_a\otimes\hat{I}_B)\hat{\rho}_{AB}(\hat{M}_a^\dagger\otimes\hat{I}_B)}{p_a}\right]
$$

**Bob's unconditional state** (averaged over Alice's outcomes, which Bob cannot observe):

$$
\hat{\rho}_B = \sum_a p_a\hat{\rho}_{B\vert a} = \sum_a \operatorname{Tr}_A\!\left[(\hat{M}_a^\dagger\hat{M}_a\otimes\hat{I}_B)\hat{\rho}_{AB}\right]
= \operatorname{Tr}_A\!\left[\left(\sum_a\hat{M}_a^\dagger\hat{M}_a\otimes\hat{I}_B\right)\hat{\rho}_{AB}\right]
$$

Using completeness $\sum_a \hat{M}_a^\dagger\hat{M}_a = \hat{I}_A$:

$$
\hat{\rho}_B = \operatorname{Tr}_A(\hat{\rho}_{AB}) = \hat{\rho}_B^{\text{initial}}
$$

**Conclusion.** Bob's marginal state is completely unaffected by what Alice measures (or whether she measures at all). Without classical communication, Bob cannot detect any sign of Alice's measurement. Entanglement creates correlations in joint outcomes but cannot transmit information. $\square$
:::

### Quantum Teleportation

Alice can transmit an unknown state $\vert\psi\rangle = \alpha\vert 0\rangle + \beta\vert 1\rangle$ to Bob using one shared Bell pair and 2 classical bits:

1. **Setup**: Alice holds qubit C (the state to send) and qubit A; Bob holds qubit B. Qubits A and B share $\vert\Phi^+\rangle_{AB} = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$.
2. **Bell measurement**: Alice measures qubits C and A in the Bell basis, obtaining 2 classical bits $m_1 m_2$.
3. **Classical communication**: Alice sends $m_1 m_2$ to Bob.
4. **Correction**: Bob applies $\hat{X}^{m_2}\hat{Z}^{m_1}$ to qubit B, recovering $\vert\psi\rangle_B$.

:::{admonition} Example: Teleportation Algebra
:class: dropdown example

**Problem.** Verify that after Alice's Bell measurement yields outcome $m_1 m_2$, Bob's correction $\hat{X}^{m_2}\hat{Z}^{m_1}$ recovers $\vert\psi\rangle$.

**Solution.** Expand the initial state $\vert\psi\rangle_C \otimes \vert\Phi^+\rangle_{AB}$ in the Bell basis of qubits C and A:

$$
\vert\psi\rangle_C \vert\Phi^+\rangle_{AB} = \tfrac{1}{2}\bigl[\vert\Phi^+\rangle_{CA}\vert\psi\rangle_B + \vert\Phi^-\rangle_{CA}\hat{Z}\vert\psi\rangle_B + \vert\Psi^+\rangle_{CA}\hat{X}\vert\psi\rangle_B + \vert\Psi^-\rangle_{CA}\hat{Z}\hat{X}\vert\psi\rangle_B\bigr]
$$

Each Bell outcome leaves Bob's qubit in a Pauli-rotated version of $\vert\psi\rangle$. The 2 bits $(m_1, m_2)$ tell Bob which Pauli to undo.
:::

:::{admonition} Why Teleportation Respects No-Communication
:class: attention

Alice's Bell measurement produces **random** outcomes. Without the 2 classical bits, Bob's qubit is maximally mixed — indistinguishable from noise. The quantum state is reconstructed only after Bob receives the classical message, which travels at most at light speed.
:::

### Summary

- **Bell's theorem**: no local hidden variable theory reproduces quantum predictions; the CHSH bound $\leq 2$ is violated by quantum mechanics at $2\sqrt{2}$.
- **Loophole-free experiments** (2015+) confirm quantum nonlocality with high confidence.
- **No-communication**: entanglement creates correlations, not signals; local statistics are independent of distant operations.
- **Quantum teleportation**: entanglement + 2 classical bits transmit an unknown quantum state without physical transport.
