## Lecture Notes

### Overview

When a quantum system is prepared in a single pure state $\vert\psi\rangle$, the system is fully characterized. But what happens when we possess only incomplete information—perhaps due to classical ignorance, or because the system has interacted with an inaccessible environment? The **density matrix** captures this, unifying the description of pure and mixed states under a single formalism.

### The Density Matrix

:::{admonition} Density Matrix
:class: important

**Pure state:** For a known state $\vert\psi\rangle$, the density operator is the rank-1 projector:

$$
\hat{\rho} = \vert\psi\rangle\langle\psi\vert
$$ (eq-mixed-pure-density)

**Mixed state (ensemble):** When the system is in state $\vert\psi_i\rangle$ with probability $p_i$ ($p_i \geq 0$, $\sum_i p_i = 1$):

$$
\hat{\rho} = \sum_i p_i \vert\psi_i\rangle\langle\psi_i\vert
$$ (eq-mixed-ensemble-density)

**Conditions for a valid density matrix:**

1. **Hermitian:** $\hat{\rho}^\dagger = \hat{\rho}$
2. **Trace one:** $\operatorname{Tr}(\hat{\rho}) = 1$
3. **Positive semidefinite:** $\hat{\rho} \geq 0$ (all eigenvalues $\lambda_i \geq 0$)

:::

Diagonal elements of $\hat{\rho}$ in a basis are **populations** (outcome probabilities); off-diagonal elements are **coherences** (interference terms).

:::{admonition} Example: Superposition vs Mixture
:class: dropdown example

**Problem.** Compare the density matrices for (i) the superposition $\vert +\rangle = (\vert 0\rangle + \vert 1\rangle)/\sqrt{2}$ and (ii) the classical mixture of $\vert 0\rangle$ and $\vert 1\rangle$ with equal probability.

**Solution.**

$$
\begin{split}
\hat{\rho}_{\text{super}} &= \frac{1}{2}\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix}, \\
\hat{\rho}_{\text{mix}} &= \frac{1}{2}\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = \frac{\hat{I}}{2}
\end{split}
$$

Both have identical diagonal elements (equal probability of measuring $\vert 0\rangle$ or $\vert 1\rangle$), but $\hat{\rho}_{\text{super}}$ has off-diagonal coherences while $\hat{\rho}_{\text{mix}}$ does not. Measuring $\hat{X}$ distinguishes them: $\langle\hat{X}\rangle = 1$ for the superposition but $0$ for the mixture.
:::

### Purity

:::{admonition} Purity
:class: important

The **purity** of a density matrix is:

$$
\mathcal{P} = \operatorname{Tr}(\hat{\rho}^2) = \sum_i \lambda_i^2
$$ (eq-mixed-purity)

where $\lambda_i$ are the eigenvalues of $\hat{\rho}$.

- **Pure state:** $\hat{\rho}^2 = \hat{\rho}$ (idempotent), so $\operatorname{Tr}(\hat{\rho}^2) = 1$.
- **Maximally mixed** ($d$-dimensional): all $\lambda_i = 1/d$, so $\operatorname{Tr}(\hat{\rho}^2) = 1/d$.
- **General:** $1/d \leq \operatorname{Tr}(\hat{\rho}^2) \leq 1$.

:::

:::{admonition} Derivation: Purity Bounds
:class: dropdown information

**Pure state:** $\hat{\rho} = \vert\psi\rangle\langle\psi\vert$ gives $\hat{\rho}^2 = \vert\psi\rangle\langle\psi\vert\psi\rangle\langle\psi\vert = \hat{\rho}$, so $\operatorname{Tr}(\hat{\rho}^2) = 1$.

**Lower bound:** The Cauchy-Schwarz inequality $(\sum_i a_i b_i)^2 \leq (\sum_i a_i^2)(\sum_i b_i^2)$, applied with $a_i = \lambda_i$ and $b_i = 1$, gives $(\sum_i \lambda_i)^2 \leq (\sum_i \lambda_i^2)(\sum_i 1) = d\sum_i \lambda_i^2$. Since $\sum_i \lambda_i = 1$, this rearranges to $\sum_i \lambda_i^2 \geq 1/d$, with equality (the maximally mixed state) when all $\lambda_i = 1/d$.

**Mixed state:** Expanding the square, $(\sum_i \lambda_i)^2 = \sum_i \lambda_i^2 + 2\sum_{i<j} \lambda_i \lambda_j$. If at least two $\lambda_i$ are positive, the cross-term $2\sum_{i<j} \lambda_i \lambda_j$ is strictly positive, hence $\sum_i \lambda_i^2 < (\sum_i \lambda_i)^2 = 1$.
:::

### Ensemble Ambiguity

:::{admonition} Ensemble Ambiguity
:class: attention

Different statistical ensembles can produce the **same** density matrix. For example, the maximally mixed qubit $\hat{\rho} = \hat{I}/2$ arises equally from mixing $\{\vert 0\rangle, \vert 1\rangle\}$ or $\{\vert +\rangle, \vert -\rangle\}$ with equal probabilities. No measurement can distinguish the two preparations.

The density matrix is the fundamental physical object—not the ensemble that generated it.
:::

Any density matrix $\hat{\rho} = \sum_i \lambda_i \vert u_i\rangle\langle u_i\vert$ has a unique **spectral decomposition** in terms of its eigenvectors $\{\vert u_i\rangle\}$. All other decompositions (with non-orthogonal states) represent different ensembles yielding the same physics.

:::{admonition} Discussion: distinguishing equivalent mixtures
:class: dropdown tip

A qubit in the state $\hat{\rho} = \hat{I}/2$ can be prepared by mixing $\vert 0\rangle$ and $\vert 1\rangle$ with equal probability, or by mixing $\vert +\rangle$ and $\vert -\rangle$—no measurement can distinguish these preparations. Yet in statistical mechanics, we routinely assign a *specific* thermal ensemble $\{p_n, \vert n\rangle\}$ to a system at temperature $T$.

- If the density matrix is the fundamental object, why do physicists still talk about specific ensembles?
- Can a system "know" how it was prepared, or does all preparation information get erased once the state is described by $\hat{\rho}$?
- How does this connect to entropy: does $S(\hat{\rho})$ measure our ignorance, or something objective about the system?
:::

### The Bloch Ball

:::{admonition} Bloch Ball Representation
:class: important

A qubit density matrix can be written as:

$$
\hat{\rho} = \frac{1}{2}(\hat{I} + \boldsymbol{r} \cdot \hat{\boldsymbol{\sigma}})
$$ (eq-mixed-bloch-density)

where $\hat{\boldsymbol{\sigma}} \equiv (\hat{X}, \hat{Y}, \hat{Z})$ is the Pauli vector (we also write $\hat{\sigma}^i$ for its $i$-th component, $i \in \{x, y, z\}$), and $\boldsymbol{r} = (\langle\hat{X}\rangle, \langle\hat{Y}\rangle, \langle\hat{Z}\rangle)$ is the **Bloch vector** satisfying $\vert\boldsymbol{r}\vert \leq 1$.

- $\vert\boldsymbol{r}\vert = 1$: pure state (Bloch sphere surface)
- $\vert\boldsymbol{r}\vert < 1$: mixed state (Bloch ball interior)
- $\boldsymbol{r} = \boldsymbol{0}$: maximally mixed ($\hat{\rho} = \hat{I}/2$, ball center)

The purity is $\operatorname{Tr}(\hat{\rho}^2) = \frac{1}{2}(1 + \vert\boldsymbol{r}\vert^2)$.

:::

:::{admonition} Derivation: Purity from Bloch Vector
:class: dropdown information

Start from $\hat{\rho} = \frac{1}{2}(\hat{I} + \boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})$. Two auxiliary identities enter the chain: the Pauli identity $(\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})^2 = \vert\boldsymbol{r}\vert^2 \hat{I}$, and the traces $\operatorname{Tr}(\hat{I}) = 2$, $\operatorname{Tr}(\hat{\sigma}^i) = 0$ for $i \in \{x, y, z\}$.

$$
\begin{split}
\operatorname{Tr}(\hat{\rho}^2) &= \frac{1}{4}\operatorname{Tr}(\hat{I} + \boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})^2 \\
&= \frac{1}{4}\operatorname{Tr}(\hat{I} + 2\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}} + (\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})^2) \\
&= \frac{1}{4}\operatorname{Tr}(\hat{I} + 2\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}} + \vert\boldsymbol{r}\vert^2 \hat{I}) \\
&= \frac{1}{4}(2 + 0 + 2\vert\boldsymbol{r}\vert^2) = \frac{1}{2}(1 + \vert\boldsymbol{r}\vert^2).
\end{split}
$$

The four steps are: square $\hat{\rho}$ and pull out $1/4$; binomially expand $(\hat{I} + \boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})^2$ without yet using the Pauli identity; substitute the Pauli identity into the squared term; apply $\operatorname{Tr}\hat{I} = 2$ to the first and last terms and $\operatorname{Tr}\hat{\sigma}^i = 0$ to kill the cross term, then simplify.

:::

### Expectation Values

:::{admonition} Expectation Value from Density Matrix
:class: important

For any observable $\hat{A}$:

$$
\langle \hat{A} \rangle = \operatorname{Tr}(\hat{\rho} \hat{A})
$$ (eq-mixed-expectation)

This formula applies to both pure and mixed states. For pure states, it reduces to $\langle\psi\vert\hat{A}\vert\psi\rangle$.
:::

### Time Evolution

:::{admonition} Von Neumann Equation
:class: important

The density matrix evolves as:

$$
\mathrm{i}\hbar\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = [\hat{H}, \hat{\rho}]
$$ (eq-mixed-von-neumann)

Equivalently, $\hat{\rho}(t) = \hat{U}(t)\hat{\rho}(0)\hat{U}^\dagger(t)$ where $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$.

:::

:::{admonition} Derivation: Von Neumann Equation
:class: dropdown information

The density matrix evolves by conjugation with the propagator, $\hat{\rho}(t) = \hat{U}(t)\,\hat{\rho}(0)\,\hat{U}^\dagger(t)$, where $\hat{\rho}(0)$ is time-independent. The propagator obeys $\frac{\mathrm{d}\hat{U}}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}\hat{H}\hat{U}$; its Hermitian conjugate $\hat{U}^\dagger(t) = \mathrm{e}^{+\mathrm{i}\hat{H}t/\hbar}$ differentiates with the opposite sign, $\frac{\mathrm{d}\hat{U}^\dagger}{\mathrm{d}t} = +\frac{\mathrm{i}}{\hbar}\hat{U}^\dagger\hat{H}$. Apply the product rule (two terms, since $\hat{\rho}(0)$ is constant), substitute both derivatives, and regroup $\hat{U}\,\hat{\rho}(0)\,\hat{U}^\dagger = \hat{\rho}(t)$:

$$
\begin{split}
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t}
&= \frac{\mathrm{d}\hat{U}}{\mathrm{d}t}\,\hat{\rho}(0)\,\hat{U}^\dagger + \hat{U}\,\hat{\rho}(0)\,\frac{\mathrm{d}\hat{U}^\dagger}{\mathrm{d}t} \\
&= -\frac{\mathrm{i}}{\hbar}\hat{H}\hat{U}\,\hat{\rho}(0)\,\hat{U}^\dagger + \frac{\mathrm{i}}{\hbar}\hat{U}\,\hat{\rho}(0)\,\hat{U}^\dagger\hat{H} \\
&= -\frac{\mathrm{i}}{\hbar}\hat{H}\,\hat{\rho}(t) + \frac{\mathrm{i}}{\hbar}\hat{\rho}(t)\,\hat{H} \\
&= -\frac{\mathrm{i}}{\hbar}\big(\hat{H}\hat{\rho}(t) - \hat{\rho}(t)\hat{H}\big) \\
&= -\frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{\rho}(t)]
\end{split}
$$

Multiplying through by $\mathrm{i}\hbar$ gives the von Neumann equation Eq. {eq}`eq-mixed-von-neumann`. The opposite signs of the two propagator derivatives are what produce the commutator structure.

:::

:::{admonition} Purity is Conserved
:class: attention

Under unitary evolution, $\operatorname{Tr}(\hat{\rho}^2)$ is constant:

$$
\operatorname{Tr}((\hat{U}\hat{\rho}\hat{U}^\dagger)^2) = \operatorname{Tr}(\hat{\rho}^2)
$$

Unitary evolution **cannot convert a pure state into a mixed state**. Apparent mixing requires interaction with an inaccessible environment (decoherence, §6.4.1).
:::

### Summary

- **Density matrix definition:** $\hat{\rho} = \sum_i p_i \vert\psi_i\rangle\langle\psi_i\vert$ (ensemble mixture); satisfies $\operatorname{Tr}(\hat{\rho}) = 1$ and $\hat{\rho} = \hat{\rho}^\dagger$ (Hermitian, normalized, positive semidefinite).

- **Purity:** $\operatorname{Tr}(\hat{\rho}^2) = 1$ iff pure ($\hat{\rho} = \vert\psi\rangle\langle\psi\vert$); less than 1 iff mixed (statistical mixture or ignorance).

- **Ensemble ambiguity:** Given $\hat{\rho}$, infinitely many ensemble decompositions $\{p_i, \vert\psi_i\rangle\}$ exist; all physical predictions are the same — the density matrix is the unique observable invariant.

- **Bloch ball:** A qubit state takes the form $\hat{\rho} = \tfrac{1}{2}(\hat{I} + \boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})$ with Bloch vector $\vert\boldsymbol{r}\vert\le 1$: surface ($\vert\boldsymbol{r}\vert = 1$) for pure states, interior for mixed, and the center $\boldsymbol{r} = \boldsymbol{0}$ for the maximally mixed state $\hat{I}/2$; purity is $(1 + \vert\boldsymbol{r}\vert^2)/2$.

- **Expectation values:** Any observable obeys $\langle\hat{A}\rangle = \operatorname{Tr}(\hat{\rho}\hat{A})$, reducing to $\langle\psi\vert\hat{A}\vert\psi\rangle$ on pure states.

- **Unitary time evolution:** $\mathrm{i}\hbar\,\partial_t\hat{\rho} = [\hat{H}, \hat{\rho}]$ (Von Neumann equation); purity is conserved under $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$, so a pure state cannot evolve into a mixed one without coupling to an inaccessible environment.
