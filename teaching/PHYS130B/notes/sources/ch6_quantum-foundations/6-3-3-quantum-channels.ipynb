## Lecture Notes

### Overview

A **quantum channel** is a linear map that describes the most general physical evolution of a quantum system. Unlike unitary evolution (which is reversible), quantum channels include irreversible processes: measurement, decoherence, noise. The key insight is that **any** physical process transforming density matrices must satisfy complete positivity (CPTP) and trace preservation. The Kraus representation provides the universal form: every CPTP channel can be written as a sum of "jump operators" $\mathcal{E}(\hat{\rho}) = \sum_k K_k \hat{\rho} K_k^\dagger$. In this section, we define CPTP maps rigorously, derive the Kraus theorem, and examine canonical examples (depolarizing, dephasing, amplitude damping) that model real quantum noise.

### Definition: Quantum Channels

A **quantum channel** (or **quantum operation**) is a linear map:

$$
\mathcal{E}: \text{Density matrices on } \mathcal{H}_\text{in} \to \text{Density matrices on } \mathcal{H}_\text{out}
$$

representing any physical process: unitary evolution, measurement, dissipation, entanglement with environment, etc.

#### CPTP Properties

A valid quantum channel must satisfy four conditions:

**1. Linearity**:

$$
\mathcal{E}(p\hat{\rho}_1 + q\hat{\rho}_2) = p\mathcal{E}(\hat{\rho}_1) + q\mathcal{E}(\hat{\rho}_2)
$$ (eq-channel-linearity)

**2. Trace preservation**:

$$
\text{Tr}(\mathcal{E}(\hat{\rho})) = \text{Tr}(\hat{\rho}) = 1
$$ (eq-channel-trace-preserving)

Probabilities are conserved; output is a valid density matrix.

**3. Positivity**:

$$
\hat{\rho} \geq 0 \quad \Rightarrow \quad \mathcal{E}(\hat{\rho}) \geq 0
$$ (eq-channel-positivity)

Positive semidefinite inputs map to positive semidefinite outputs.

**4. Complete positivity**:

$$
\mathcal{E} \otimes \mathcal{I}_n \text{ is positive for all } n
$$ (eq-channel-completely-positive)

Even when the system is entangled with auxiliary systems, the map preserves positivity.

:::{admonition} Physical Motivation for Complete Positivity
:class: note

Consider two qubits A and B, entangled. You apply a channel $\mathcal{E}$ to A only. If $\mathcal{E}$ is merely positive but **not** completely positive, the map $\mathcal{E} \otimes \mathcal{I}_B$ could produce **negative eigenvalues** in the joint system (negative probability!). This is unphysical. Therefore, all physical channels must be completely positive.
:::

:::{admonition} Definition: CPTP Channel
:class: important

A **completely positive, trace-preserving (CPTP) map** $\mathcal{E}$ is a linear map on density matrices satisfying linearity, trace preservation, positivity, and complete positivity. CPTP maps are the mathematical characterization of all physical quantum processes.
:::

### Kraus Representation Theorem

**Theorem**: A linear map $\mathcal{E}$ on density matrices is CPTP if and only if it can be written in **Kraus form**:

$$
\mathcal{E}(\hat{\rho}) = \sum_{k=1}^r K_k \hat{\rho} K_k^\dagger
$$ (eq-channel-kraus)

where the **Kraus operators** $\{K_k\}$ satisfy:

$$
\sum_k K_k^\dagger K_k = I
$$ (eq-channel-kraus-constraint)

The Kraus representation ensures complete positivity and traces preservation:

:::{admonition} Proof: Kraus Representation Satisfies CPTP
:class: dropdown information

**Trace-preserving**: 
$$\text{Tr}(\sum_k K_k \hat{\rho} K_k^\dagger) = \sum_k \text{Tr}(\hat{\rho} K_k^\dagger K_k) = \text{Tr}(\hat{\rho} \sum_k K_k^\dagger K_k) = \text{Tr}(\hat{\rho} \cdot I) = \text{Tr}(\hat{\rho})$$

**Positive semidefinite output**: If $\hat{\rho} \geq 0$, then $\mathcal{E}(\hat{\rho}) = \sum_k K_k \hat{\rho} K_k^\dagger \geq 0$ (sum of positive operators).

**Completely positive**: The Kraus structure ensures positivity even under tensor product with identity (standard proof involves expanding in eigenbasis).
:::

#### Non-Uniqueness of Kraus Operators

Given one set $\{K_k\}$, other valid sets exist related by unitary rotation:

$$
K_k' = \sum_j U_{kj} K_j
$$

where $\{U_{kj}\}$ is unitary. Different Kraus representations encode the same physical channel.

:::{admonition} Derivation: Kraus Theorem via Choi–Jamiołkowski Isomorphism
:class: dropdown information

**Goal:** Prove that *every* CPTP map $\mathcal{E}$ has a Kraus representation $\mathcal{E}(\hat{\rho}) = \sum_k K_k\hat{\rho}K_k^\dagger$ with $\sum_k K_k^\dagger K_k = \hat{I}$.

**Step 1: The maximally entangled reference state.** Introduce an auxiliary reference system $R$ with the same dimension $d$ as $S$. Define the (unnormalized) maximally entangled state:

$$
\vert\Omega\rangle_{SR} = \sum_{i=1}^d \vert i\rangle_S \otimes \vert i\rangle_R \in \mathcal{H}_S \otimes \mathcal{H}_R,
$$

so that $\vert\Omega\rangle\langle\Omega\vert_{SR} = \sum_{i,j}\vert i\rangle\langle j\vert_S\otimes\vert i\rangle\langle j\vert_R$.

**Step 2: The Choi matrix.** Apply $\mathcal{E}$ to the $S$ part only, leaving $R$ untouched:

$$
C_\mathcal{E} \;=\; (\mathcal{E}_S \otimes \mathcal{I}_R)\bigl(\vert\Omega\rangle\langle\Omega\vert_{SR}\bigr) \;=\; \sum_{i,j} \mathcal{E}(\vert i\rangle\langle j\vert)_S \otimes \vert i\rangle\langle j\vert_R.
$$

The **Choi matrix** $C_\mathcal{E}$ is a $d^2 \times d^2$ matrix acting on $\mathcal{H}_S\otimes\mathcal{H}_R$.

**Step 3: Complete positivity $\Leftrightarrow$ Choi matrix is positive semidefinite.** Since $\mathcal{E}$ is completely positive, the map $\mathcal{E}\otimes\mathcal{I}_R$ is positive; therefore

$$
C_\mathcal{E} = (\mathcal{E}\otimes\mathcal{I}_R)(\vert\Omega\rangle\langle\Omega\vert) \geq 0.
$$

(This is the Choi–Jamiołkowski isomorphism: $\mathcal{E}$ is CP $\Leftrightarrow$ $C_\mathcal{E}\geq 0$.)

**Step 4: Spectral decomposition.** Since $C_\mathcal{E}\geq 0$, it has a spectral decomposition:

$$
C_\mathcal{E} = \sum_k \lambda_k \vert v_k\rangle\langle v_k\vert, \qquad \lambda_k \geq 0,\quad \vert v_k\rangle \in \mathcal{H}_S\otimes\mathcal{H}_R.
$$

**Step 5: Reshape eigenvectors into Kraus operators.** Each eigenvector can be written as

$$
\vert v_k\rangle = \sum_{i,j}(K_k)_{ij}\,\vert i\rangle_S \otimes \vert j\rangle_R,
$$

which defines a $d\times d$ matrix $K_k$ (the $\vert i\rangle_S\otimes\vert j\rangle_R$ coefficient is $(K_k)_{ij}$). Define rescaled operators $K_k \leftarrow \sqrt{\lambda_k}K_k$.

**Step 6: Recover the channel.** By the reshaping isomorphism, for any $d\times d$ matrix $\hat{\rho}$:

$$
\mathcal{E}(\hat{\rho}) = \sum_{i,j}\hat{\rho}_{ij}\,\mathcal{E}(\vert i\rangle\langle j\vert) = \sum_k K_k\hat{\rho}K_k^\dagger.
$$

**Step 7: Trace preservation → completeness relation.** Since $\mathcal{E}$ is trace-preserving, $\operatorname{Tr}_S(C_\mathcal{E}) = \hat{I}_R$:

$$
\operatorname{Tr}_S(C_\mathcal{E}) = \sum_{i,j}\operatorname{Tr}(\mathcal{E}(\vert i\rangle\langle j\vert))\,\vert i\rangle\langle j\vert_R = \sum_{i,j}\delta_{ij}\,\vert i\rangle\langle j\vert_R = \hat{I}_R.
$$

In terms of Kraus operators, $\operatorname{Tr}_S(C_\mathcal{E}) = \sum_k K_k^\dagger K_k$, so $\sum_k K_k^\dagger K_k = \hat{I}$. $\square$

**Summary:** Any CPTP map is in bijection with a positive Choi matrix via the Choi–Jamiołkowski isomorphism. The Kraus operators are the (reshaped) eigenvectors of the Choi matrix. The completeness condition $\sum_k K_k^\dagger K_k = \hat{I}$ follows from trace preservation.
:::

:::{admonition} Derivation: Amplitude Damping Channel from Spontaneous Emission
:class: dropdown information

We derive the Kraus operators for the **amplitude damping channel**, which describes spontaneous emission: a two-level atom in the excited state $\vert 1\rangle$ decays to the ground state $\vert 0\rangle$ with probability $\gamma$, emitting a photon.

**Stinespring dilation.** Model the system (atom) + environment (photon modes initially in vacuum $\vert 0\rangle_E$). The most general single-photon-emission interaction takes the form of a unitary $\hat{U}$ on $\mathcal{H}_S \otimes \mathcal{H}_E$:

$$
\hat{U}(\vert 0\rangle_S \otimes \vert 0\rangle_E) = \vert 0\rangle_S \otimes \vert 0\rangle_E
$$
$$
\hat{U}(\vert 1\rangle_S \otimes \vert 0\rangle_E) = \sqrt{1-\gamma}\vert 1\rangle_S \otimes \vert 0\rangle_E + \sqrt{\gamma}\vert 0\rangle_S \otimes \vert 1\rangle_E
$$

The ground state $\vert 0\rangle$ is stable; the excited state $\vert 1\rangle$ either remains with probability $1-\gamma$ or decays (emitting a photon into the environment) with probability $\gamma$.

**Action on a general state.** For $\vert\psi\rangle_S = \alpha\vert 0\rangle + \beta\vert 1\rangle$:

$$
\hat{U}(\vert\psi\rangle_S \otimes \vert 0\rangle_E) = \alpha\vert 0\rangle_S\vert 0\rangle_E + \beta\bigl(\sqrt{1-\gamma}\vert 1\rangle_S\vert 0\rangle_E + \sqrt{\gamma}\vert 0\rangle_S\vert 1\rangle_E\bigr)
$$

**Extract Kraus operators.** The Kraus operators are defined by $K_j = {}_E\langle j\vert \hat{U} \vert 0\rangle_E$ (project environment onto each basis state):

$$
K_0 = {}_E\langle 0\vert \hat{U} \vert 0\rangle_E:\quad K_0\vert 0\rangle = \vert 0\rangle, \quad K_0\vert 1\rangle = \sqrt{1-\gamma}\vert 1\rangle \implies K_0 = \begin{pmatrix} 1 & 0 \\ 0 & \sqrt{1-\gamma} \end{pmatrix}
$$

$$
K_1 = {}_E\langle 1\vert \hat{U} \vert 0\rangle_E:\quad K_1\vert 0\rangle = 0, \quad K_1\vert 1\rangle = \sqrt{\gamma}\vert 0\rangle \implies K_1 = \begin{pmatrix} 0 & \sqrt{\gamma} \\ 0 & 0 \end{pmatrix}
$$

**Completeness check:**

$$
K_0^\dagger K_0 + K_1^\dagger K_1 = \begin{pmatrix}1&0\\0&1-\gamma\end{pmatrix} + \begin{pmatrix}0&0\\0&\gamma\end{pmatrix} = \begin{pmatrix}1&0\\0&1\end{pmatrix} = \hat{I} \quad \checkmark
$$

**Channel action on density matrix.** For $\hat{\rho} = \begin{pmatrix}\rho_{00}&\rho_{01}\\\rho_{10}&\rho_{11}\end{pmatrix}$:

$$
\mathcal{E}_{\text{AD}}(\hat{\rho}) = K_0\hat{\rho}K_0^\dagger + K_1\hat{\rho}K_1^\dagger = \begin{pmatrix}\rho_{00}+\gamma\rho_{11} & \sqrt{1-\gamma}\rho_{01}\\ \sqrt{1-\gamma}\rho_{10} & (1-\gamma)\rho_{11}\end{pmatrix}
$$

**Physical interpretation.** The excited-state population $\rho_{11}$ decays to zero; the ground-state population $\rho_{00}$ increases by $\gamma\rho_{11}$; coherences $\rho_{01}$ are damped by $\sqrt{1-\gamma}$ per step. This is the $T_1$ (longitudinal) relaxation process in NMR and qubit physics.
:::

:::

### Common Quantum Channels

#### 1. Unitary Channel

$$
\mathcal{E}_U(\hat{\rho}) = U \hat{\rho} U^\dagger
$$ (eq-channel-unitary)

**Kraus form**: Single operator $K = U$ with $U^\dagger U = I$.

**Properties**: Reversible (inverse is $\mathcal{E}_{U^\dagger}$); entropy conserved; no information loss.

#### 2. Depolarizing Channel

With probability $p$, the state becomes completely random; with probability $1-p$, unchanged:

$$
\mathcal{E}_\text{depol}(\hat{\rho}) = (1-p)\hat{\rho} + p\frac{I}{d}
$$ (eq-channel-depolarizing)

where $d = \dim(\mathcal{H})$.

**Kraus operators** (qubit, $d=2$):

$$
K_0 = \sqrt{1-p}\, I, \quad K_1 = \sqrt{p/3}\, \hat{\sigma}^x, \quad K_2 = \sqrt{p/3}\, \hat{\sigma}^y, \quad K_3 = \sqrt{p/3}\, \hat{\sigma}^z
$$ (eq-channel-depol-kraus)


:::{admonition} Derivation: Depolarizing Channel in Kraus Form
:class: dropdown information

We verify that the four Kraus operators $K_0 = \sqrt{1-p}\,\hat{I}$, $K_{1,2,3} = \sqrt{p/3}\,\hat{\sigma}^{x,y,z}$ reproduce the depolarizing channel $\mathcal{E}(\hat{\rho}) = (1-p)\hat{\rho} + (p/d)\hat{I}$.

**Step 1: Verify completeness.** $\sum_{k=0}^3 K_k^\dagger K_k = (1-p)\hat{I} + \frac{p}{3}(\hat{\sigma}^x)^2 + \frac{p}{3}(\hat{\sigma}^y)^2 + \frac{p}{3}(\hat{\sigma}^z)^2 = (1-p)\hat{I} + \frac{p}{3}\cdot 3\hat{I} = \hat{I}$. ✓

**Step 2: Compute the channel action.** Write $\hat{\rho} = \frac{1}{2}(\hat{I} + \boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})$. Using $\hat{\sigma}^a\hat{\sigma}^b\hat{\sigma}^a = -\hat{\sigma}^b$ for $a \neq b$ and $\hat{\sigma}^a\hat{\sigma}^a\hat{\sigma}^a = \hat{\sigma}^a$:

$$
K_0\hat{\rho}K_0^\dagger = (1-p)\hat{\rho}
$$

$$
K_a\hat{\rho}K_a^\dagger = \frac{p}{3}\hat{\sigma}^a\hat{\rho}\hat{\sigma}^a = \frac{p}{3}\cdot\frac{1}{2}(\hat{I} - r_a\hat{\sigma}^a - r_b\hat{\sigma}^b - r_c\hat{\sigma}^c + \ldots) \quad (a\neq b,c)
$$

Summing over $a = x,y,z$: each Pauli component $\hat{\sigma}^b$ receives $(-1)$ from two of the three Pauli conjugations and $(+1)$ from one, giving net $-1/3$ per component. Thus:

$$
\sum_{a=x,y,z} \frac{p}{3}\hat{\sigma}^a\hat{\rho}\hat{\sigma}^a = \frac{p}{2}\hat{I} - \frac{p}{3}\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}}
$$

**Step 3: Total.** Combining:

$$
\mathcal{E}(\hat{\rho}) = (1-p)\hat{\rho} + \frac{p}{2}\hat{I} - \frac{p}{3}\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}} = \left(1-p-\frac{p}{3}\right)\frac{\boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}}}{2} + \frac{\hat{I}}{2} + \frac{p}{2}\hat{I} - \frac{\hat{I}}{2}
$$

Collecting: $\mathcal{E}(\hat{\rho}) = (1-\frac{4p}{3})\hat{\rho} + \frac{p}{3}\hat{I}$, which for $p \in [0, 3/4]$ shrinks the Bloch vector by factor $(1-4p/3)$ toward the maximally mixed state. This is the standard depolarizing channel. $\square$
:::

**Physical interpretation**: Errors on all three Pauli axes with equal probability; models generic noise.

#### 3. Amplitude Damping ($T_1$ Process)

Excited state decays to ground state at rate $\Gamma = 1/T_1$:

$$
\mathcal{E}_\text{amp}(\hat{\rho}) = K_0 \hat{\rho} K_0^\dagger + K_1 \hat{\rho} K_1^\dagger
$$ (eq-channel-amplitude-damping)

with:

$$
K_0 = \begin{pmatrix} 1 & 0 \\ 0 & \sqrt{1-\gamma} \end{pmatrix}, \quad K_1 = \begin{pmatrix} 0 & \sqrt{\gamma} \\ 0 & 0 \end{pmatrix}
$$ (eq-channel-amp-kraus)

where $\gamma = 1 - \mathrm{e}^{-t/T_1}$ is the decay probability.

**Physical interpretation**: Energy dissipation; ground state is absorbing (fixed point).

#### 4. Phase Damping ($T_2$ Process, Dephasing)

Coherences decay without energy loss at rate $\Gamma_\phi = 1/T_2$:

$$
\mathcal{E}_\text{phase}(\hat{\rho}) = K_0 \hat{\rho} K_0^\dagger + K_1 \hat{\rho} K_1^\dagger
$$ (eq-channel-phase-damping)

with:

$$
K_0 = \begin{pmatrix} 1 & 0 \\ 0 & \sqrt{1-\gamma} \end{pmatrix}, \quad K_1 = \begin{pmatrix} 0 & 0 \\ 0 & \sqrt{\gamma} \end{pmatrix}
$$ (eq-channel-phase-kraus)

where $\gamma = 1 - \mathrm{e}^{-t/T_2}$.

**Physical interpretation**: Random phase fluctuations; off-diagonal elements (coherences) decay; diagonal (populations) preserved.

### Reversibility and Irreversibility

**Reversible (Unitary) Channel**:
- Single Kraus operator $K = U$ (unitary).
- Entropy conserved: $S(\mathcal{E}(\hat{\rho})) = S(\hat{\rho})$.
- Inverse channel exists: $\mathcal{E}^{-1}(\sigma) = U^\dagger \sigma U$.
- State can be recovered perfectly.

**Irreversible (Non-Unitary) Channel**:
- Multiple Kraus operators ($r > 1$) or degenerate structure.
- Entropy generally increases: $S(\mathcal{E}(\hat{\rho})) > S(\hat{\rho})$ (output more mixed).
- No inverse: original state cannot be recovered (information lost).
- Examples: Depolarizing, amplitude damping.

:::{admonition} Derivation: Stinespring Dilation — Every Channel Is a Unitary on a Larger Space
:class: dropdown information

The **Stinespring dilation theorem** shows that every CPTP map arises from a unitary evolution on a larger Hilbert space followed by a partial trace. This makes precise the idea that "open system" dynamics is secretly "closed system" dynamics with extra degrees of freedom.

**Theorem.** For any CPTP map $\mathcal{E}$ on $\mathcal{H}_S$, there exists an environment Hilbert space $\mathcal{H}_E$, an initial environment state $\vert E_0\rangle$, and a unitary $U$ on $\mathcal{H}_S \otimes \mathcal{H}_E$ such that:

$$
\mathcal{E}(\hat{\rho}) = \mathrm{Tr}_E\!\left[U\,\bigl(\hat{\rho} \otimes \vert E_0\rangle\langle E_0\vert\bigr)\,U^\dagger\right]
$$

**Construction from Kraus operators.**

Suppose $\mathcal{E}(\hat{\rho}) = \sum_{k=1}^r K_k \hat{\rho} K_k^\dagger$ with $\sum_k K_k^\dagger K_k = I$.

Let $\{\vert E_k\rangle\}_{k=1}^r$ be an orthonormal basis for an environment of dimension $r$. Define the **Stinespring isometry** $V: \mathcal{H}_S \to \mathcal{H}_S \otimes \mathcal{H}_E$ by:

$$
V\vert\psi\rangle = \sum_{k=1}^r K_k\vert\psi\rangle \otimes \vert E_k\rangle
$$

**Step 1: $V$ is an isometry ($V^\dagger V = I_S$).**

$$
V^\dagger V \vert\psi\rangle = \sum_{j,k} \langle E_j\vert E_k\rangle\, K_j^\dagger K_k \vert\psi\rangle = \sum_k K_k^\dagger K_k\vert\psi\rangle = I_S\vert\psi\rangle
$$

using orthonormality $\langle E_j\vert E_k\rangle = \delta_{jk}$ and the Kraus completeness condition.

**Step 2: Partial trace recovers the channel.**

Starting from the joint state $V\hat{\rho}V^\dagger = \sum_{k,k'} K_k\hat{\rho}K_{k'}^\dagger \otimes \vert E_k\rangle\langle E_{k'}\vert$, tracing over the environment:

$$
\mathrm{Tr}_E(V\hat{\rho}V^\dagger) = \sum_{j}\langle E_j\vert\left(\sum_{k,k'} K_k\hat{\rho}K_{k'}^\dagger \otimes \vert E_k\rangle\langle E_{k'}\vert\right)\vert E_j\rangle = \sum_k K_k\hat{\rho}K_k^\dagger = \mathcal{E}(\hat{\rho})
$$

**Step 3: Extending to a unitary.**

The isometry $V$ maps $\mathcal{H}_S$ into a subspace of $\mathcal{H}_S \otimes \mathcal{H}_E$. It can always be extended to a full unitary $U$ on $\mathcal{H}_S \otimes \mathcal{H}_E$ (by choosing an arbitrary unitary extension on the orthogonal complement). Setting $\vert E_0\rangle$ to any fixed state mapped to the initial environment, the formula $\mathcal{E}(\hat{\rho}) = \mathrm{Tr}_E[U(\hat{\rho}\otimes\vert E_0\rangle\langle E_0\vert)U^\dagger]$ follows.

**Converse direction.**

Conversely, given any unitary $U$ and initial environment state $\vert E_0\rangle$, define $K_k = \langle E_k\vert U\vert E_0\rangle$ (operators on $\mathcal{H}_S$). Then:

$$
\mathrm{Tr}_E\left[U(\hat{\rho}\otimes\vert E_0\rangle\langle E_0\vert)U^\dagger\right] = \sum_k K_k\hat{\rho}K_k^\dagger
$$

with $\sum_k K_k^\dagger K_k = \langle E_0\vert U^\dagger U\vert E_0\rangle = I_S$. Every such construction is CPTP.

**Physical interpretation.** At the fundamental level, all quantum dynamics is unitary — the apparent irreversibility of open-system evolution comes entirely from discarding (tracing over) the environment. The Stinespring theorem makes this precise: every noisy channel has a "secret" unitary implementation.
:::

### Summary

- **Quantum channel**: CPTP linear map from density matrices to density matrices; describes all physical quantum evolution.
- **CPTP**: Completely positive and trace-preserving; ensures valid quantum states and conserved probabilities.
- **Kraus representation** $\mathcal{E}(\hat{\rho}) = \sum_k K_k \hat{\rho} K_k^\dagger$ with $\sum_k K_k^\dagger K_k = I$: universal form for channels; non-unique.
- **Common channels**: Unitary (reversible), depolarizing (generic noise), amplitude damping ($T_1$), phase damping ($T_2$).
- **Reversibility**: Unitary channels conserve entropy; non-unitary channels increase entropy (information loss).
- **Unification**: Unitary evolution is a special case ($r=1$); projective measurement is a special case (see [6.3.2 POVM](6-3-2-povm)).

- **Phase-flip** (Z noise): $\mathcal{E}_Z(\hat{\rho}) = (1-p)\hat{\rho} + p\sigma^z\hat{\rho}\sigma^z$
- **Depolarizing**: $\mathcal{E}_\text{depol}(\hat{\rho}) = (1-p)\hat{\rho} + \frac{p}{2}I$ (random Pauli, normalized)

(a) Apply each channel to the input state $\hat{\rho}_\text{in} = \frac{1}{2}(I + \hat{\sigma}^x)$ (eigenstate of $\sigma^x$). Compare the outputs: which channel preserves the most information?

(b) **Conceptual**: The depolarizing channel is called "universal noise" because it cannot distinguish the computational basis. Explain why bit-flip and phase-flip channels **can** be distinguished by an appropriate measurement on the output, but depolarizing cannot.
