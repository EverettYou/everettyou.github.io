## Lecture Notes

When a qubit couples to its environment—a thermal bath, stray photons, or surrounding atoms—the environment becomes entangled with the system and effectively "measures" it. The off-diagonal elements of the density matrix decay, turning a coherent superposition into a classical mixture. This is **decoherence**: the mechanism behind the quantum-to-classical transition.

### Overview

Decoherence can be understood in two steps: the environment *entangles* with the system (like a measurement), then we *trace out* the environment (like forgetting the outcome). No new postulate is needed — decoherence is simply the dephasing quantum channel (§6.3.3) applied by the environment. Pointer states are the robust states that survive this process; all other superpositions decohere on a timescale $T_2 = 1/\gamma$.

:::{admonition} Discussion
:class: dropdown tip

For a qubit coupled to a thermal bath, the decoherence time satisfies $T_2 \leq 2T_1$. This bound is saturated by pure dephasing. Express $T_2$ in terms of the relaxation rate $\Gamma_1 = 1/T_1$ and pure dephasing rate $\Gamma_\varphi = 1/T_\varphi$: the relation is $\frac{1}{T_2} = \frac{1}{2T_1} + \frac{1}{T_\varphi}$. Why does energy relaxation contribute to dephasing even without random phase kicks?
:::

### Decoherence Process

Consider a qubit $S$ initially in a superposition, coupled to environment $E$ (bath of many modes). After interaction, the total state evolves to:

$$
\alpha\vert 0\rangle_S \vert E_0\rangle_E + \beta\vert 1\rangle_S \vert E_1\rangle_E \;\longrightarrow\; \alpha\vert 0\rangle_S \vert E_0(t)\rangle_E + \beta\vert 1\rangle_S \vert E_1(t)\rangle_E
$$ (eq-decoherence-entangled-state)

where $\vert E_0(t)\rangle$ and $\vert E_1(t)\rangle$ are (nearly) orthogonal environmental states. The environment has become a **which-path detector**: it encodes information about which branch the system is in.

:::{admonition} Pointer States and Einselection
:class: important

**Pointer states** are the eigenstates of the system-environment interaction Hamiltonian $\hat{H}_\text{int}$. They are the states robust under decoherence: the environment correlates with them without creating superpositions between them.

**Einselection** (Zurek): The environment *selects* a preferred pointer basis. Superpositions of pointer states decohere rapidly into mixtures; individual pointer states remain stable.

**Example**: For $\hat{H}_\text{int} = \hat{\sigma}^z \otimes B_E$, the pointer basis is $\{\vert 0\rangle, \vert 1\rangle\}$ (eigenstates of $\hat{\sigma}^z$). The superposition $\vert+\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$ decoheres rapidly; $\vert 0\rangle$ and $\vert 1\rangle$ do not.
:::

### Reduced Density Matrix

Tracing out the environment gives the **reduced density matrix**:

$$
\hat{\rho}_S(t) = \text{Tr}_E\!\left(\vert\Psi(t)\rangle\langle\Psi(t)\vert\right)
$$ (eq-decoherence-reduced)

Computing explicitly:

$$
\hat{\rho}_S(t) = \vert\alpha\vert^2 \vert 0\rangle\langle 0\vert + \vert\beta\vert^2 \vert 1\rangle\langle 1\vert + \alpha^*\beta\,\langle E_1(t)\vert E_0(t)\rangle\,\vert 0\rangle\langle 1\vert + \beta^*\alpha\,\langle E_0(t)\vert E_1(t)\rangle\,\vert 1\rangle\langle 0\vert
$$ (eq-decoherence-reduced-explicit)

As the environmental states become orthogonal ($\langle E_0\vert E_1\rangle \to 0$), the off-diagonal coherences vanish. The system transitions from a pure state to a mixed state — not from any measurement postulate, but purely from entanglement with the environment.


:::{admonition} Derivation: Reduced Density Matrix Shows Decoherence
:class: dropdown information

Starting from the entangled state $\vert\Psi(t)\rangle = \alpha\vert 0\rangle_S\vert E_0(t)\rangle_E + \beta\vert 1\rangle_S\vert E_1(t)\rangle_E$, compute the full density matrix:

$$
\hat{\rho}_{SE}(t) = \vert\alpha\vert^2\vert 0\rangle\langle 0\vert_S \otimes \vert E_0\rangle\langle E_0\vert_E + \alpha^*\beta\,\vert 1\rangle\langle 0\vert_S \otimes \vert E_0\rangle\langle E_1\vert_E + \beta^*\alpha\,\vert 0\rangle\langle 1\vert_S \otimes \vert E_1\rangle\langle E_0\vert_E + \vert\beta\vert^2\vert 1\rangle\langle 1\vert_S \otimes \vert E_1\rangle\langle E_1\vert_E
$$

**Partial trace over the environment.** Insert $\sum_n\vert\xi_n\rangle_E\langle\xi_n\vert_E = \hat{I}_E$ (orthonormal environment basis):

$$
\hat{\rho}_S(t) = \operatorname{Tr}_E\hat{\rho}_{SE}(t) = \sum_n\langle\xi_n\vert_E\hat{\rho}_{SE}(t)\vert\xi_n\rangle_E
$$

The diagonal blocks give:

$$
\hat{\rho}_S^{00}(t) = \vert\alpha\vert^2\vert 0\rangle\langle 0\vert\underbrace{\langle E_0\vert E_0\rangle}_{1} = \vert\alpha\vert^2\vert 0\rangle\langle 0\vert
$$

The off-diagonal (coherence) block:

$$
\hat{\rho}_S^{01}(t) = \alpha\beta^*\vert 0\rangle\langle 1\vert\underbrace{\langle E_1(t)\vert E_0(t)\rangle}_{= D(t)}
$$

where $D(t) = \langle E_1(t)\vert E_0(t)\rangle$ is the **decoherence factor** — the overlap between the two environmental branches.

**Result:**

$$
\hat{\rho}_S(t) = \begin{pmatrix} \vert\alpha\vert^2 & \alpha\beta^*\,D(t) \\ \alpha^*\beta\,D^*(t) & \vert\beta\vert^2 \end{pmatrix}
$$

As $D(t) \to 0$ (environment becomes orthogonal), the off-diagonal coherences vanish: $\hat{\rho}_S \to \vert\alpha\vert^2\vert 0\rangle\langle 0\vert + \vert\beta\vert^2\vert 1\rangle\langle 1\vert$ — a classical mixture. **Decoherence = environment becoming a which-branch detector.** No measurement postulate needed; it follows directly from entanglement and partial trace.
:::

### Decoherence Timescale

In the **weak coupling limit**, the environmental overlap decays exponentially:

$$
\langle E_0(t) \vert E_1(t)\rangle \approx \mathrm{e}^{-t/T_2}
$$ (eq-decoherence-exponential)

where $T_2 = 1/\gamma$ is the **dephasing time**. The microscopic derivation:

:::{admonition} Derivation: Dephasing Rate from Spin-Bath Coupling
:class: dropdown information

For $H_\text{int} = \hat{\sigma}^z \otimes \sum_k g_k (a_k \mathrm{e}^{-\mathrm{i}\omega_k t} + a_k^\dagger \mathrm{e}^{\mathrm{i}\omega_k t})$ (spin coupled to bosonic bath), the overlap integral gives:

$$
\langle E_0(t)\vert E_1(t)\rangle = \exp\!\left(-\int_0^\infty \frac{J(\omega)}{\omega^2}(1 - \cos\omega t)\coth\frac{\hbar\omega}{2k_BT}\,\mathrm{d}\omega\right)
$$

where $J(\omega) = \sum_k g_k^2\,\delta(\omega - \omega_k)$ is the spectral density of the bath. For an ohmic bath at high temperature $k_BT \gg \hbar\omega_c$, this gives exponential decay $\mathrm{e}^{-\gamma t}$ with rate $\gamma \propto k_BT\,\sum_k g_k^2/\omega_k^2$.
:::

**Macroscopic objects**: The decoherence rate scales extensively with system size $N$: $\gamma_\text{total} \sim N\gamma_\text{atom}$. For a macroscopic object ($N \sim 10^{26}$), $T_2 \sim 10^{-26}\,\text{s}$ — decoherence is essentially instantaneous. This is why we never observe macroscopic superpositions.

### Decoherence as a Quantum Channel

Decoherence is not a new phenomenon — it is the **dephasing channel** (§6.3.3) derived from a microscopic model:

1. **Entanglement** (environment measures): $|\psi\rangle \otimes |E_0\rangle \to \alpha|0\rangle|E_0(t)\rangle + \beta|1\rangle|E_1(t)\rangle$
2. **Partial trace** (forgetting): $\text{Tr}_E(\cdot)$ discards the measurement outcome.

The result is exactly the dephasing channel:

$$
\mathcal{E}(\hat{\rho}) = \begin{pmatrix} \hat{\rho}_{00} & \hat{\rho}_{01}\,\mathrm{e}^{-t/T_2} \\\ \hat{\rho}_{10}\,\mathrm{e}^{-t/T_2} & \hat{\rho}_{11} \end{pmatrix}
$$ (eq-decoherence-dephasing-channel)

This identification — decoherence = quantum channel — means the full machinery of §6.3.3 applies to open-system evolution without any new framework.

:::{admonition} Derivation: Random Phase Kicks Produce the Dephasing Channel
:class: dropdown information

A simple stochastic model: at each time step, the qubit receives a random phase kick $\hat{U}(\phi) = \mathrm{e}^{\mathrm{i}\phi\hat{Z}/2}$, where $\phi$ is drawn from a Gaussian with zero mean and variance $\sigma^2$.

**Single-kick action on density matrix.** For $\hat\rho = \begin{pmatrix}\rho_{00}&\rho_{01}\\\rho_{10}&\rho_{11}\end{pmatrix}$:

$$
\hat{U}(\phi)\hat{\rho}\hat{U}^\dagger(\phi) = \begin{pmatrix}\rho_{00} & \mathrm{e}^{\mathrm{i}\phi}\rho_{01} \\ \mathrm{e}^{-\mathrm{i}\phi}\rho_{10} & \rho_{11}\end{pmatrix}
$$

(The diagonal elements are unchanged; only the coherences acquire a phase factor.)

**Averaging over the random kick.** The channel is the ensemble average over $\phi \sim \mathcal{N}(0, \sigma^2)$:

$$
\mathcal{E}(\hat{\rho}) = \int_{-\infty}^{+\infty} \frac{\mathrm{e}^{-\phi^2/(2\sigma^2)}}{\sqrt{2\pi\sigma^2}} \hat{U}(\phi)\hat{\rho}\hat{U}^\dagger(\phi)\,\mathrm{d}\phi
$$

For the off-diagonal element $\rho_{01}$, this gives:

$$
\mathcal{E}(\hat{\rho})_{01} = \rho_{01} \int_{-\infty}^{+\infty} \frac{\mathrm{e}^{-\phi^2/(2\sigma^2)}}{\sqrt{2\pi\sigma^2}} \mathrm{e}^{\mathrm{i}\phi}\,\mathrm{d}\phi = \rho_{01} \cdot \mathrm{e}^{-\sigma^2/2}
$$

where the last step uses the Gaussian characteristic function $\langle\mathrm{e}^{\mathrm{i}k\phi}\rangle_{\mathcal{N}(0,\sigma^2)} = \mathrm{e}^{-k^2\sigma^2/2}$ with $k=1$.

**Result: dephasing channel.**

$$
\mathcal{E}(\hat{\rho}) = \begin{pmatrix}\rho_{00} & \mathrm{e}^{-\sigma^2/2}\rho_{01}\\ \mathrm{e}^{-\sigma^2/2}\rho_{10} & \rho_{11}\end{pmatrix}
$$

Identifying $\mathrm{e}^{-\sigma^2/2} = \mathrm{e}^{-t/T_2}$ relates the noise variance $\sigma^2$ to the dephasing time $T_2 = 2/\sigma^2$ per step. Accumulated over $N$ kicks: $\sigma^2_\text{total} = N\sigma^2$, so decoherence builds up linearly in time.

**Kraus representation.** The Gaussian average can be written as a Kraus sum with infinitely many operators $K_\phi = \sqrt{P(\phi)}\hat{U}(\phi)$. However, the same channel also has a two-Kraus representation: $K_0 = \sqrt{p}\hat{I}$, $K_1 = \sqrt{1-p}\hat{Z}$ (phase-flip channel), with $p = (1 + \mathrm{e}^{-\sigma^2/2})/2$, confirming the identification with the phase-flip / dephasing channel.
:::

:::{admonition} Decoherence is not measurement collapse
:class: caution

Decoherence explains why off-diagonal elements of $\hat{\rho}_S$ decay, making interference unobservable. But the full system+environment state remains **pure** — decoherence does not select a single outcome. The measurement problem (why we observe definite results) is a separate philosophical question.
:::

### Quantum-to-Classical Transition

Under decoherence, a superposition of pointer states rapidly becomes an effective classical mixture:

$$
\hat{\rho}_S(t) \approx \begin{pmatrix} 1/2 & \mathrm{e}^{-t/T_2}/2 \\\ \mathrm{e}^{-t/T_2}/2 & 1/2 \end{pmatrix} \xrightarrow{t \gg T_2} \frac{I}{2}
$$

For macroscopic objects, this happens in $\sim 10^{-26}$ s, far below any observable timescale. **This is why classical physics works**: quantum coherence is perpetually destroyed by environmental coupling before it can manifest at macroscopic scales.

### Decoherence-Free Subspaces

Some states are **immune** to specific decoherence channels due to symmetry. For two qubits coupling identically to the environment ($H_\text{int} = (\hat{\sigma}^z_1 + \hat{\sigma}^z_2)\otimes B_E$), the singlet state is decoherence-free:

$$
(\hat{\sigma}^z_1 + \hat{\sigma}^z_2)\,\vert\Psi^-\rangle = 0, \qquad \vert\Psi^-\rangle = \frac{1}{\sqrt{2}}(\vert 01\rangle - \vert 10\rangle)
$$

The singlet's total $z$-spin is zero, so the common-mode dephasing leaves it unchanged. Decoherence-free subspaces are a foundation of quantum error correction (§6.4.3).

:::{admonition} Derivation: Decoherence-Free Subspace Criterion
:class: dropdown information

A subspace $\mathcal{S}$ is **decoherence-free** if every state $\vert\psi\rangle \in \mathcal{S}$ evolves unitarily despite coupling to the environment. We derive the necessary and sufficient condition.

**Setup.** Model the system–environment interaction via Lindblad operators $\{L_k\}$ (§6.4.2). The master equation is:

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] + \sum_k \left(L_k \hat{\rho} L_k^\dagger - \frac{1}{2}\{L_k^\dagger L_k, \hat{\rho}\}\right)
$$

For a state $\vert\psi\rangle$ to evolve unitarily (no decoherence), the dissipator $\mathcal{D}[\hat{\rho}]$ must vanish on $\hat{\rho} = \vert\psi\rangle\langle\psi\vert$.

**Step 1: Necessary condition — DFS criterion.**

The dissipator $\mathcal{D}$ vanishes on $\hat{\rho} = \vert\psi\rangle\langle\psi\vert$ if and only if each $L_k$ maps $\vert\psi\rangle$ to a multiple of itself:

$$
L_k\vert\psi\rangle = \lambda_k^{(\psi)}\vert\psi\rangle \quad \text{for all } k
$$

**Proof of necessity.** Compute $\operatorname{Tr}(\hat{\rho}\,\mathcal{D}[\hat{\rho}])$ with $\hat{\rho} = \vert\psi\rangle\langle\psi\vert$:

$$
\operatorname{Tr}(\hat{\rho}\,\mathcal{D}[\hat{\rho}]) = \sum_k \operatorname{Tr}\!\left(\hat{\rho} L_k\hat{\rho}L_k^\dagger - \tfrac{1}{2}\hat{\rho}^2 L_k^\dagger L_k - \tfrac{1}{2} L_k^\dagger L_k \hat{\rho}^2\right)
$$

Since $\hat{\rho}^2 = \hat{\rho}$ for a pure state:

$$
= \sum_k \left[\langle\psi\vert L_k\vert\psi\rangle\langle\psi\vert L_k^\dagger\vert\psi\rangle - \langle\psi\vert L_k^\dagger L_k\vert\psi\rangle\right] = \sum_k \left[\vert\langle\psi\vert L_k\vert\psi\rangle\vert^2 - \langle L_k^\dagger L_k\rangle\right]
$$

By the Cauchy–Schwarz inequality, $\vert\langle\psi\vert L_k\vert\psi\rangle\vert^2 \leq \langle\psi\vert L_k^\dagger L_k\vert\psi\rangle$ with equality iff $L_k\vert\psi\rangle \propto \vert\psi\rangle$. For $\mathcal{D}[\hat{\rho}] = 0$, the quantity must be zero for every $k$, requiring $L_k\vert\psi\rangle = \lambda_k\vert\psi\rangle$.

**Step 2: Consistency for a subspace.**

For the entire subspace $\mathcal{S}$ (not just a single state) to be DFS, the eigenvalue $\lambda_k$ must be **the same** for all states in $\mathcal{S}$:

$$
L_k\vert\psi\rangle = \lambda_k\vert\psi\rangle \quad \forall\, \vert\psi\rangle \in \mathcal{S}, \; \forall\, k
$$

i.e., each $L_k$ acts as a **scalar** $\lambda_k \hat{I}$ on $\mathcal{S}$.

**Proof.** Suppose $L_k\vert\psi_1\rangle = \lambda_k^{(1)}\vert\psi_1\rangle$ and $L_k\vert\psi_2\rangle = \lambda_k^{(2)}\vert\psi_2\rangle$. For the superposition $\vert\psi_+\rangle = (\vert\psi_1\rangle + \vert\psi_2\rangle)/\sqrt{2}$:

$$
L_k\vert\psi_+\rangle = \frac{\lambda_k^{(1)}\vert\psi_1\rangle + \lambda_k^{(2)}\vert\psi_2\rangle}{\sqrt{2}}
$$

For this to be $\propto \vert\psi_+\rangle$, we need $\lambda_k^{(1)} = \lambda_k^{(2)} \equiv \lambda_k$.

**Step 3: Application to the dephasing example.**

For common-mode dephasing with $L = \hat{\sigma}^z_1 + \hat{\sigma}^z_2$, the singlet $\vert\Psi^-\rangle = (\vert01\rangle - \vert10\rangle)/\sqrt{2}$ satisfies:

$$
(\hat{\sigma}^z_1 + \hat{\sigma}^z_2)\vert 01\rangle = (1-1)\vert 01\rangle + ... \;\to\; L\vert\Psi^-\rangle = 0 \cdot \vert\Psi^-\rangle
$$

So $\lambda = 0$ for the singlet, confirming it forms a DFS with $L_k\vert\psi\rangle = 0$ (the null eigenvalue case). Any superposition of states in this one-dimensional DFS is stable: no information is leaked to the environment.

**DFS Criterion Summary.** A subspace $\mathcal{S}$ is decoherence-free if and only if the Lindblad operators act as scalars on $\mathcal{S}$:

$$
\boxed{L_k\big\vert_{\mathcal{S}} = \lambda_k \hat{I}_{\mathcal{S}} \quad \forall\, k}
$$

This is a symmetry condition: the environment "cannot distinguish" states within $\mathcal{S}$, so no which-state information leaks out and no decoherence occurs.
:::

### Summary

- Decoherence arises when the system entangles with its environment; the partial trace turns a pure state into a mixed state by destroying off-diagonal coherences.
- The coherence factor $\langle E_0(t)\vert E_1(t)\rangle = \mathrm{e}^{-t/T_2}$ decays on the dephasing timescale $T_2 = 1/\gamma$.
- Decoherence = measurement by the environment + forgetting the outcome; it is equivalent to the dephasing quantum channel (§6.3.3).
- Pointer states are einselected by the environment — they are the eigenstates of $H_\text{int}$ and remain stable under decoherence.
- Macroscopic objects decohere in $\lesssim 10^{-26}$ s, making quantum coherence unobservable at human scales.
- Decoherence-free subspaces exploit symmetry to protect quantum information from specific noise channels.
