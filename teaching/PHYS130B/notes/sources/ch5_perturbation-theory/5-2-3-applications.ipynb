## Lecture Notes

### Overview

This lesson applies first-order time-dependent perturbation theory to four key physical systems: harmonic driving that reveals resonance and Rabi oscillations, electric dipole transitions with selection rules, decay and linewidth from finite lifetime, and linear response theory leading to the quantum Hall conductance—a remarkable exact result from perturbation theory.

### Harmonic Perturbation

The simplest time-dependent perturbation is a harmonic (oscillating) drive:

$$
V(t) = V_0 \cos(\omega t)
$$

We can rewrite this as a sum of positive and negative frequency components:

$$
V(t) = \frac{V_0}{2} \left( \mathrm{e}^{\mathrm{i}\omega t} + \mathrm{e}^{-\mathrm{i}\omega t} \right)
$$

The transition amplitude from state $\vert i \rangle$ to $\vert f \rangle$ picks up resonance when the driving frequency matches the energy gap:

$$
c_f^{(1)}(t) \propto \langle f | V_0 | i \rangle \left[ \frac{\mathrm{e}^{\mathrm{i}(\omega - \omega_{fi})t} - 1}{\omega - \omega_{fi}} + \frac{\mathrm{e}^{-\mathrm{i}(\omega + \omega_{fi})t} - 1}{\omega + \omega_{fi}} \right]
$$

where $\omega_{fi} = (E_f - E_i)/\hbar$. The first term dominates when $\omega \approx \omega_{fi}$ (absorption), and the second when $\omega \approx -\omega_{fi}$ (stimulated emission). This is the rotating wave approximation: we keep only the resonant term and drop the rapidly oscillating counter-rotating term.

Near resonance, the transition probability oscillates at the **Rabi frequency**:

:::{admonition} Rabi Oscillations
:class: important

In the rotating frame near resonance, the transition probability is

$$
P(t) = \sin^2\left(\frac{\Omega_R t}{2}\right) \quad \text{where} \quad \Omega_R = \frac{|V_{fi}|}{\hbar}
$$

is the Rabi frequency. Population coherently oscillates between the two states at rate $\Omega_R$. A $\pi$-pulse ($\Omega_R t = \pi$) completely flips the state.

:::


:::{admonition} Derivation: Rabi Oscillations from the Two-Level System
:class: dropdown information

Consider a two-level system $\{\vert i\rangle, \vert f\rangle\}$ driven by a harmonic perturbation. In the rotating frame (interaction picture with the drive frequency $\omega$), the Hamiltonian reduces to a time-independent $2\times 2$ matrix. Writing $\delta = \omega_{fi} - \omega$ for the detuning and $\Omega_R = \vert V_{fi}\vert/\hbar$ for the on-resonance Rabi frequency:

$$
\hat{H}_{\text{RWA}} = \frac{\hbar}{2}\begin{pmatrix} \delta & \Omega_R \\ \Omega_R & -\delta \end{pmatrix}
$$

(assuming real $V_{fi}$; this is the **rotating wave approximation** — the counter-rotating term oscillating at $2\omega$ is dropped as non-resonant).

**Eigenvalues.** The generalized Rabi frequency is $\Omega = \sqrt{\delta^2 + \Omega_R^2}$, giving eigenvalues $\pm\hbar\Omega/2$.

**Transition probability.** Starting in $\vert i\rangle = (1, 0)^\top$ at $t = 0$, the exact time evolution gives:

$$
P_{i\to f}(t) = \frac{\Omega_R^2}{\Omega^2}\sin^2\!\left(\frac{\Omega t}{2}\right)
$$

**On resonance** ($\delta = 0$, $\Omega = \Omega_R$): $P_{i\to f}(t) = \sin^2(\Omega_R t/2)$ — complete population oscillation at frequency $\Omega_R$. A $\pi$-pulse ($\Omega_R t = \pi$) achieves full inversion.

**Off resonance** ($\delta \neq 0$): the frequency increases to $\Omega > \Omega_R$ but the amplitude $\Omega_R^2/\Omega^2 < 1$. As $|\delta| \to \infty$, the probability is suppressed as $(\Omega_R/\delta)^2 \ll 1$.

**Connection to first-order perturbation theory.** For weak coupling and short times ($\Omega_R t \ll 1$, $\delta \approx 0$): $\sin^2(\Omega_R t/2) \approx \Omega_R^2 t^2/4 = \vert V_{fi}\vert^2 t^2/(4\hbar^2)$, reproducing the quadratic short-time growth of first-order TDPT.
:::

A key application: **scanning the driving frequency $\omega$ probes the energy level spacing**. Each resonance reveals $E_f - E_i$, forming the basis of spectroscopy.

### Electric Dipole Transitions

When light couples to matter, the interaction is dominated by the dipole term:

$$
V(t) = -\boldsymbol{d} \cdot \boldsymbol{E}(t) = -(-e\boldsymbol{r}) \cdot \boldsymbol{E}(t) = e \boldsymbol{r} \cdot \boldsymbol{E}(t)
$$

where $\boldsymbol{d} = -e\boldsymbol{r}$ is the electric dipole moment. The transition amplitude depends on the dipole matrix element $\langle f | \boldsymbol{r} | i \rangle$.

Not all transitions are allowed. The parity and angular momentum structure of wavefunctions impose **selection rules**. For hydrogen (or any single-electron atom):

:::{admonition} Selection Rules for Electric Dipole Transitions
:class: important

For a transition $n\ell m_\ell \to n'\ell' m_{\ell}'$:
- $\Delta \ell = \pm 1$ (change parity)
- $\Delta m_\ell = 0, \pm 1$ (depends on light polarization)
- $\Delta S = 0$ (spin unchanged in non-relativistic theory)

Transitions violating these are forbidden and have suppressed amplitude. States that can only decay via forbidden transitions become **metastable** (long-lived).

:::

For example, the 2s state of hydrogen cannot decay to 1s by electric dipole radiation (both $\ell = 0$, so $\Delta \ell = 0$), making 2s metastable. The 2p state, however, can decay resonantly.

### Lifetime and Linewidth

An excited state coupled to a continuum of final states (e.g., radiating photons) decays exponentially. From Fermi's golden rule applied to the emission process, the decay rate is

$$
\Gamma = \frac{2\pi}{\hbar} \sum_f |\langle f | V | i \rangle|^2 \delta(E_i - E_f - \hbar\omega_{\text{photon}})
$$

The inverse is the **lifetime**:

:::{admonition} Lifetime and Natural Linewidth
:class: important

$$
\tau = \frac{1}{\Gamma}
$$

The energy-time uncertainty principle implies that a state with finite lifetime $\tau$ has an intrinsic energy uncertainty:

$$
\Delta E \sim \hbar \Gamma = \frac{\hbar}{\tau}
$$

This gives a **natural linewidth** in the emission spectrum. The lineshape is Lorentzian, centered at the transition energy $E_i - E_f$ with width $\Gamma$.

:::


:::{admonition} Derivation: Lorentzian Lineshape from Exponential Decay
:class: dropdown information

When an excited state $\vert i\rangle$ decays at rate $\Gamma$, the Wigner–Weisskopf theory models its amplitude as:

$$
c_i(t) = \mathrm{e}^{-\mathrm{i}E_i t/\hbar}\,\mathrm{e}^{-\Gamma t/2}, \qquad t \geq 0
$$

so that the population $\vert c_i(t)\vert^2 = \mathrm{e}^{-\Gamma t} = \mathrm{e}^{-t/\tau}$ decays with lifetime $\tau = 1/\Gamma$.

**Energy spectrum.** The Fourier transform gives the amplitude at energy $E$:

$$
\tilde{c}_i(E) = \int_0^\infty \mathrm{e}^{\mathrm{i}(E-E_i)t/\hbar}\,\mathrm{e}^{-\Gamma t/2}\,\mathrm{d}t = \frac{\mathrm{i}\hbar}{E - E_i + \mathrm{i}\hbar\Gamma/2}
$$

The spectral probability density (lineshape) is:

$$
\vert\tilde{c}_i(E)\vert^2 \propto \frac{1}{(E - E_i)^2 + (\hbar\Gamma/2)^2}
$$

This is a **Lorentzian** centered at $E_i$ with full width at half-maximum:

$$
\Delta E_{\text{FWHM}} = \hbar\Gamma = \frac{\hbar}{\tau}
$$

This is the **natural linewidth** — a consequence of the energy-time uncertainty $\Delta E\cdot\tau \sim \hbar$. In the limit $\Gamma \to 0$ (infinite lifetime), the Lorentzian narrows to $\delta(E - E_i)$, recovering the sharp spectral line of a stationary state.
:::

For instance, the 2p state of hydrogen has $\tau \sim 1.6 \times 10^{-9}$ s (lifetime), corresponding to $\Gamma \sim 0.1$ eV (linewidth).
:::{admonition} Derivation: Einstein A and B Coefficients from Thermal Equilibrium
:class: dropdown information

Consider a two-level atom with ground state $\vert g\rangle$ and excited state $\vert e\rangle$ (energy gap $\hbar\omega_0$) in thermal equilibrium with radiation at temperature $T$.

**Three radiative processes** occur simultaneously:

1. **Absorption**: atom absorbs a photon, jumping $\vert g\rangle\to\vert e\rangle$. Rate $= B_{ge}\,u(\omega_0)$, where $u(\omega_0)$ is the spectral energy density.
2. **Stimulated emission**: a photon triggers $\vert e\rangle\to\vert g\rangle$ (emitting a second identical photon). Rate $= B_{eg}\,u(\omega_0)$.
3. **Spontaneous emission**: $\vert e\rangle\to\vert g\rangle$ without external radiation, rate $= A_{eg}$ (independent of $u$).

**Detailed balance.** In thermal equilibrium the Boltzmann ratio fixes populations:

$$
\frac{N_e}{N_g} = \mathrm{e}^{-\hbar\omega_0/k_{\mathrm{B}}T}
$$

Setting the total upward rate equal to the total downward rate:

$$
N_g\,B_{ge}\,u(\omega_0) = N_e\bigl[A_{eg} + B_{eg}\,u(\omega_0)\bigr]
$$

Solving for $u(\omega_0)$:

$$
u(\omega_0) = \frac{A_{eg}}{B_{ge}\,\mathrm{e}^{\hbar\omega_0/k_{\mathrm{B}}T} - B_{eg}}
$$

**Matching Planck's law.** The Planck distribution at frequency $\omega_0$ is:

$$
u(\omega_0) = \frac{\hbar\omega_0^3}{\pi^2 c^3}\,\frac{1}{\mathrm{e}^{\hbar\omega_0/k_{\mathrm{B}}T}-1}
$$

For the two expressions to agree **at every temperature** $T$, we need both the exponential coefficient and the constant term to match:

$$
B_{ge} = B_{eg} \equiv B \qquad \text{(stimulated rates are equal)}
$$

$$
\frac{A_{eg}}{B} = \frac{\hbar\omega_0^3}{\pi^2 c^3} \qquad \text{(Einstein relation)}
$$

**Why $B_{ge} = B_{eg}$.** This follows directly from Fermi's golden rule: since $\vert\langle e\vert\hat{V}\vert g\rangle\vert^2 = \vert\langle g\vert\hat{V}\vert e\rangle\vert^2$ by Hermiticity, absorption and stimulated emission have identical matrix elements. Population inversion (needed for lasing) requires external pumping to push the system away from equilibrium.

**The $A/B$ ratio and vacuum fluctuations.** The spontaneous emission rate grows as $\omega_0^3$:

$$
A_{eg} = \frac{\hbar\omega_0^3}{\pi^2 c^3}\,B
$$

Physically, $A/B$ is proportional to the electromagnetic mode density $\rho(\omega_0) = \omega_0^2/(\pi^2 c^3)$: spontaneous emission is stimulated emission driven by **vacuum fluctuations** in those modes. Placing the atom in a cavity that suppresses (or enhances) modes at $\omega_0$ changes $A$ — this is the **Purcell effect**.

**Connection to Fermi's golden rule.** Applying FGR to the dipole coupling $\hat{V} = -\hat{\boldsymbol{d}}\cdot\boldsymbol{E}$ and averaging over polarization directions:

$$
B = \frac{\pi}{3\hbar^2\epsilon_0}\bigl\vert\langle e\vert\hat{\boldsymbol{d}}\vert g\rangle\bigr\vert^2
$$

Substituting into the Einstein relation gives the spontaneous emission rate:

$$
A_{eg} = \frac{\omega_0^3}{3\pi\hbar\epsilon_0 c^3}\bigl\vert\langle e\vert\hat{\boldsymbol{d}}\vert g\rangle\bigr\vert^2
$$

For hydrogen $2p\to 1s$: $\omega_0\approx 1.55\times 10^{16}$ rad/s and $\vert\langle\hat{d}\rangle\vert\approx 0.86\,ea_0$, giving $\tau = 1/A_{eg}\approx 1.6$ ns, consistent with the measured lifetime quoted above.
:::


### Linear Response and Hall Conductance

An electric field $\boldsymbol{E}(t)$ drives currents in the material. For weak fields, the response is **linear**:

$$
j_\alpha(t) = \sum_\beta \sigma_{\alpha\beta} E_\beta(t)
$$

where $\sigma_{\alpha\beta}$ is the conductivity tensor. Computing $\sigma$ from first-order perturbation theory requires summing the current-current correlation:

:::{admonition} Kubo Formula for Conductivity
:class: important

$$
\sigma_{xy} = \frac{2\pi e^2}{\hbar} \sum_{n \neq m} \frac{f(E_n) - f(E_m)}{(E_n - E_m)^2} \cdot 2\operatorname{Im}\langle n | \hat{j}_x | m \rangle \langle m | \hat{j}_y | n \rangle
$$

where $f(E)$ is the Fermi-Dirac distribution and $\hat{j}$ is the current operator. For $T = 0$, the sum is over occupied ($n$) and empty ($m$) states.

:::

:::{admonition} Quantum Hall Conductance from Filled Landau Levels
:class: note

When a Landau level is completely filled (all $N_\phi$ states occupied, where $N_\phi = B A / \phi_0$ is the number of states), the Kubo formula yields an exact, quantized result:

$$
\sigma_{xy} = \frac{\nu e^2}{h}
$$

where $\nu$ is the filling factor (number of filled Landau levels). For a single filled level, $\sigma_{xy} = e^2/h$—**exact** to all orders in perturbation theory, independent of disorder or interactions.

This connects back to §4.3 (Landau levels and cyclotron motion): the quantum Hall effect is not just a band structure curiosity—it is a fundamental linear response result. Perturbation theory reveals its quantization.

:::


:::{admonition} Discussion: Why Is the Hall Conductance Exactly Quantized?
:class: dropdown tip

- The conductance is independent of disorder, sample geometry, and interaction strength. Why does perturbation theory capture an exact result at all?
- The integer $\nu$ counts filled Landau levels. What is the physical meaning of this number from the perspective of topology and winding numbers?
- If we gradually deform the Hamiltonian without closing a gap, can $\sigma_{xy}$ change? What happens if we allow the gap to close?

:::


:::{admonition} Derivation: Kubo Formula from Linear Response Theory
:class: dropdown information

Apply a weak electric field $E_\beta(t) = E_\beta\,\mathrm{e}^{-\mathrm{i}\omega t+\epsilon t}$ (adiabatically switched on from $-\infty$). The dipole perturbation in the interaction picture is:

$$
\hat{V}(t) = -e\hat{r}_\beta E_\beta\,\mathrm{e}^{-\mathrm{i}\omega t+\epsilon t}
$$

**First-order current response.** From the Kubo formula of linear response theory, the change in the current expectation value is:

$$
\delta\langle\hat{j}_\alpha\rangle = \frac{1}{\mathrm{i}\hbar}\int_{-\infty}^t \langle[\hat{j}_\alpha(t),\hat{V}(t')]\rangle_0\,\mathrm{d}t'
$$

Inserting a complete set of states $\sum_m\vert m\rangle\langle m\vert$ and using $\hat{j}_\alpha = -e\hat{v}_\alpha$, and $\langle n\vert\hat{v}_\alpha\vert m\rangle = \mathrm{i}(E_m-E_n)\langle n\vert\hat{r}_\alpha\vert m\rangle/\hbar$:

$$
\sigma_{\alpha\beta}(\omega) = \frac{\mathrm{i}e^2}{\hbar}\sum_n f(E_n)\sum_{m\neq n}\frac{\langle n\vert\hat{v}_\alpha\vert m\rangle\langle m\vert\hat{v}_\beta\vert n\rangle}{\omega_{mn}+\omega+\mathrm{i}\epsilon} + \text{h.c.}
$$

**Hall conductance at $T=0$.** Taking $\omega\to 0$ and the imaginary part relevant for $\sigma_{xy}$:

$$
\sigma_{xy} = \frac{2\pi e^2}{\hbar}\sum_{n\text{ filled}}\sum_{m\text{ empty}}\frac{2\operatorname{Im}\langle n\vert\hat{v}_x\vert m\rangle\langle m\vert\hat{v}_y\vert n\rangle}{(E_n-E_m)^2}
$$

**Berry curvature form.** Using $\langle n\vert\hat{v}_\alpha\vert m\rangle = (E_m-E_n)\langle n\vert\partial_{k_\alpha}\vert m\rangle/\hbar$ (from the Bloch Hamiltonian) and the identity for the Berry curvature $F_{xy}^{(n)} = -2\operatorname{Im}\sum_{m\neq n}\langle n\vert\partial_{k_x}\hat{H}\vert m\rangle\langle m\vert\partial_{k_y}\hat{H}\vert n\rangle/(E_m-E_n)^2$:

$$
\sigma_{xy} = \frac{e^2}{h}\cdot\frac{1}{2\pi}\sum_{\text{filled bands}}\int_{\text{BZ}}F_{xy}^{(n)}(\boldsymbol{k})\,\mathrm{d}^2k = \frac{e^2}{h}\,\nu
$$

The Brillouin-zone integral of the Berry curvature is $2\pi$ times the **Chern number** $\nu$ — an integer by topology. This makes $\sigma_{xy}$ **exactly quantized** and immune to disorder or interactions as long as the bulk gap is open.
:::


:::{admonition} Derivation: Optical Bloch Equations — Coherent Drive with Spontaneous Decay
:class: dropdown information

The Rabi solution (§5.2.3) describes a coherently driven two-level system with no decay. Real atoms also undergo spontaneous emission at rate $\Gamma = A_{21}$ (Einstein $A$ coefficient). Combining both gives the **optical Bloch equations**.

**Bloch vector.** Represent the density matrix of the two-level system as a Bloch vector $\boldsymbol{r} = (r_x, r_y, r_z) = (\text{Re}\rho_{ge}, \text{Im}\rho_{ge}, \rho_{ee} - \rho_{gg})$. In the rotating frame with detuning $\Delta = \omega - \omega_0$:

$$
\dot{r}_x = \Delta\, r_y - \frac{\Gamma}{2} r_x,
$$

$$
\dot{r}_y = -\Delta\, r_x + \Omega\, r_z - \frac{\Gamma}{2} r_y,
$$

$$
\dot{r}_z = -\Omega\, r_y - \Gamma(r_z + 1).
$$

Here $\Omega = q|E_0|\langle g|z|e\rangle/\hbar$ is the Rabi frequency and $\Gamma$ enters as the transverse decay rate $T_1 = 1/\Gamma$ (population) and the coherence decay rate $T_2 = 2/\Gamma$ (pure radiative limit with $T_2 = 2T_1$).

**Steady state.** Setting $\dot{\boldsymbol{r}} = 0$ and solving:

$$
r_z^{(\mathrm{ss})} = -\frac{1 + (2\Delta/\Gamma)^2}{1 + (2\Delta/\Gamma)^2 + 2\Omega^2/\Gamma^2}.
$$

**Saturation.** Define the saturation parameter $s = 2\Omega^2/\Gamma^2$. At resonance ($\Delta = 0$):

$$
r_z^{(\mathrm{ss})} = -\frac{1}{1+s}.
$$

For $s \gg 1$ (strong drive), $r_z \to 0$: equal population in ground and excited states — saturation. The population difference $\rho_{ee} = (1 - r_z)/2 \to 1/2$.

**Power-broadened linewidth.** The steady-state excited population $\rho_{ee}(\Delta) = \frac{s/2}{1+s+(2\Delta/\Gamma)^2}$ is a Lorentzian in $\Delta$ with **power-broadened** half-width

$$
\Delta_{1/2} = \frac{\Gamma}{2}\sqrt{1+s}.
$$

At strong drive, the linewidth grows as $\sqrt{s} \propto \Omega$: the spectral line broadens beyond the natural linewidth $\Gamma$, a direct consequence of saturation.
:::

### Summary

- **Harmonic driving** at frequency $\omega$ induces resonant transitions when $\omega = \omega_{fi}$; Rabi oscillations probe the coupling strength.
- **Dipole selection rules** ($\Delta \ell = \pm 1$, etc.) determine which transitions are allowed; metastable states bypass these rules.
- **Finite lifetime** $\tau = 1/\Gamma$ produces natural linewidth $\Delta E \sim \hbar/\tau$; spectral lines have intrinsic width.
- **Linear response** and the Kubo formula predict exact Hall quantization for filled Landau levels—a triumph of first-order perturbation theory.
