## Lecture Notes

### Overview

Most quantum systems cannot be solved exactly. **Perturbation theory** provides a systematic approximation: start from a solvable Hamiltonian $H_0$, add a small perturbation $\lambda V$, and expand eigenvalues and eigenstates as power series in $\lambda$. The qubit model $H(\lambda) = \hat{\sigma}^z + \lambda\hat{\sigma}^x$ is the ideal testing ground — a $2\times 2$ system with an exact solution that we can Taylor-expand and compare order-by-order.

### The Qubit Hamiltonian

$$

H(\lambda) = \hat{\sigma}^z + \lambda\hat{\sigma}^x = \begin{pmatrix} 1 & \lambda \\ \lambda & -1 \end{pmatrix}

$$ (eq-5a-qubit-ham)

Here $H_0 = \hat{\sigma}^z$ is the unperturbed Hamiltonian with eigenstates $|0\rangle$ (energy $+1$) and $|1\rangle$ (energy $-1$), and $V = \hat{\sigma}^x$ induces spin flips between them.

### Exact Solution

Diagonalizing:

$$

\det(H - E\hat{I}) = (1-E)(-1-E) - \lambda^2 = 0 \quad\Rightarrow\quad E^2 = 1 + \lambda^2

$$

:::{admonition} Exact Eigenvalues
:class: important

$$

E_\pm(\lambda) = \pm\sqrt{1 + \lambda^2}

$$ (eq-5a-exact-energies)

The energy gap $\Delta E = 2\sqrt{1+\lambda^2}$ never closes for real $\lambda$ — the two levels **repel** rather than cross.
:::

:::{admonition} Derivation: Exact Eigenstates
:class: dropdown information

To find the eigenstate $|\psi_+\rangle$ with eigenvalue $E_+ = \sqrt{1+\lambda^2}$, solve $(H - E_+\hat{I})|\psi\rangle = 0$:

$$

\begin{pmatrix} 1 - \sqrt{1+\lambda^2} & \lambda \\ \lambda & -1 - \sqrt{1+\lambda^2} \end{pmatrix} \begin{pmatrix} a \\ b \end{pmatrix} = 0

$$

From the first row: $(1 - \sqrt{1+\lambda^2})\,a + \lambda\,b = 0$, giving

$$

\frac{b}{a} = \frac{\sqrt{1+\lambda^2} - 1}{\lambda}

$$

Setting $a = 1 + \sqrt{1+\lambda^2}$ and $b = \lambda$ (clearing the denominator), the norm is

$$

\mathcal{N}^2 = (1+\sqrt{1+\lambda^2})^2 + \lambda^2 = 2(1+\lambda^2) + 2\sqrt{1+\lambda^2} = 2(1+\lambda^2+\sqrt{1+\lambda^2})

$$

so the normalized eigenstate is

$$

|\psi_+(\lambda)\rangle = \frac{(1+\sqrt{1+\lambda^2})|0\rangle + \lambda|1\rangle}{\sqrt{2(1+\lambda^2+\sqrt{1+\lambda^2})}}

$$

The other eigenstate $|\psi_-\rangle$ follows from orthogonality (swap $|0\rangle\leftrightarrow|1\rangle$ and flip the sign of $\lambda$):

$$

|\psi_-(\lambda)\rangle = \frac{(1+\sqrt{1+\lambda^2})|1\rangle - \lambda|0\rangle}{\sqrt{2(1+\lambda^2+\sqrt{1+\lambda^2})}}

$$

At $\lambda = 0$: $|\psi_+\rangle = |0\rangle$ and $|\psi_-\rangle = |1\rangle$, as expected.
:::

### Taylor Expansion: Recovering the Perturbative Series

Expanding $\sqrt{1+\lambda^2}$ for small $\lambda$:

$$

E_\pm(\lambda) = \pm\left(1 + \frac{\lambda^2}{2} - \frac{\lambda^4}{8} + O(\lambda^6)\right)

$$ (eq-5a-energy-series)

Reading off the perturbative coefficients for the upper state:

$$

E_+^{(0)} = 1, \quad E_+^{(1)} = 0, \quad E_+^{(2)} = \frac{1}{2}, \quad E_+^{(3)} = 0, \quad E_+^{(4)} = -\frac{1}{8}

$$

All **odd-order corrections vanish** because $H(\lambda)$ has a $\lambda \to -\lambda$ symmetry (equivalently, $\hat{\sigma}^x$ has zero diagonal elements in the $|0\rangle, |1\rangle$ basis, so $E_n^{(1)} = \langle n|\hat{\sigma}^x|n\rangle = 0$).

The exact eigenstates expanded to low order:

$$

|\psi_+(\lambda)\rangle = |0\rangle + \frac{\lambda}{2}|1\rangle - \frac{\lambda^2}{8}|0\rangle + O(\lambda^3)

$$ (eq-5a-state-series-plus)

The admixture of $|1\rangle$ grows linearly with $\lambda$, controlled by the ratio coupling/gap.

:::{admonition} Derivation: Binomial Series for $\sqrt{1+\lambda^2}$
:class: dropdown information

Apply the generalized binomial series $(1+x)^\alpha = \sum_{k=0}^\infty \binom{\alpha}{k} x^k$ with $\alpha = 1/2$ and $x = \lambda^2$:

$$

\sqrt{1+\lambda^2} = 1 + \tfrac{1}{2}\lambda^2 + \tfrac{1}{2!}\tfrac{1}{2}(\tfrac{1}{2}-1)\lambda^4 + \tfrac{1}{3!}\tfrac{1}{2}(\tfrac{1}{2}-1)(\tfrac{1}{2}-2)\lambda^6 + \cdots

$$

Computing each coefficient:

$$

k=1:\quad \binom{1/2}{1} = \tfrac{1}{2}, \qquad k=2:\quad \binom{1/2}{2} = \tfrac{(1/2)(1/2-1)}{2!} = -\tfrac{1}{8}

$$

$$

k=3:\quad \binom{1/2}{3} = \tfrac{(1/2)(-1/2)(-3/2)}{3!} = \tfrac{1}{16}

$$

So $E_\pm(\lambda) = \pm(1 + \tfrac{1}{2}\lambda^2 - \tfrac{1}{8}\lambda^4 + \tfrac{1}{16}\lambda^6 - \cdots)$, confirming $E_+^{(2)} = +\tfrac{1}{2}$ and $E_+^{(4)} = -\tfrac{1}{8}$.

For the eigenstates, expand $|\psi_+(\lambda)\rangle$ by substituting $\sqrt{1+\lambda^2} = 1 + \tfrac{1}{2}\lambda^2 + \cdots$ into the exact formula and Taylor-expanding the normalization factor $\mathcal{N}^{-1}$. At the numerator level:

$$

(1+\sqrt{1+\lambda^2})|0\rangle + \lambda|1\rangle = (2 + \tfrac{1}{2}\lambda^2 + \cdots)|0\rangle + \lambda|1\rangle

$$

$$

\mathcal{N}^2 = 2(1+\lambda^2+\sqrt{1+\lambda^2}) = 2(2 + \tfrac{3}{2}\lambda^2 + \cdots) = 4(1 + \tfrac{3}{4}\lambda^2 + \cdots)

$$

$$

\mathcal{N}^{-1} = \tfrac{1}{2}(1 - \tfrac{3}{8}\lambda^2 + \cdots)

$$

Combining: $|\psi_+\rangle = |0\rangle + \tfrac{\lambda}{2}|1\rangle - \tfrac{\lambda^2}{8}|0\rangle + O(\lambda^3)$, matching the perturbative result.
:::

### Level Repulsion

:::{admonition} Level Repulsion
:class: important

Two energy levels coupled by a perturbation $V$ repel each other rather than crossing. The minimum gap at any $\lambda$ is proportional to the matrix element $|\langle 1|V|0\rangle|$. In the qubit model, $\langle 1|\hat{\sigma}^x|0\rangle = 1$, so the levels never touch.

This is a general phenomenon: for a $2\times 2$ block with off-diagonal coupling $v$, the eigenvalues are $E_\pm = \bar{E} \pm \sqrt{\delta^2 + v^2}$, where $\delta$ is the half-gap and $\bar{E}$ the midpoint. The minimum splitting $2|v|$ occurs at $\delta = 0$.
:::

:::{admonition} Derivation: Level Repulsion from Second-Order Perturbation Theory
:class: dropdown information

Level repulsion can be derived two ways: (i) from the exact 2×2 eigenvalue formula, and (ii) from the perturbative expansion. Both illuminate why coupling prevents level crossing.

**Method 1: Exact 2×2 argument.**
For any Hermitian $2\times 2$ matrix with energies $E_0 > E_1$ (half-gap $\delta = (E_0-E_1)/2 > 0$) and off-diagonal coupling $v$ (real for simplicity):

$$
\hat{H} = \bar{E}\hat{I} + \begin{pmatrix} \delta & v \\ v & -\delta \end{pmatrix}, \qquad E_\pm = \bar{E} \pm \sqrt{\delta^2 + v^2}
$$

The eigenvalue gap is $\Delta E = 2\sqrt{\delta^2 + v^2} \geq 2\delta$, with equality only when $v = 0$. For any non-zero coupling, the gap **strictly exceeds** the unperturbed gap $2\delta$. Levels can only cross if both $v = 0$ and $\delta = 0$ simultaneously — a codimension-2 condition that requires tuning two parameters.

**Method 2: Second-order perturbation theory.**
Apply non-degenerate perturbation theory to $\hat{H} = \hat{H}_0 + \lambda\hat{V}$ with $\hat{H}_0 = \hat{\sigma}^z$, $\hat{V} = \hat{\sigma}^x$. Unperturbed: $E_0^{(0)} = +1$, $E_1^{(0)} = -1$, with $\hat{V}_{10} = \langle 1|\hat{\sigma}^x|0\rangle = 1$.

**First-order corrections:** $E_n^{(1)} = \langle n|\hat{V}|n\rangle = \langle n|\hat{\sigma}^x|n\rangle = 0$ (diagonal Pauli matrix elements vanish). No first-order energy shift.

**Second-order corrections:**

$$
E_0^{(2)} = \frac{|\langle 1|\lambda\hat{V}|0\rangle|^2}{E_0^{(0)} - E_1^{(0)}} = \frac{\lambda^2 \cdot 1}{1 - (-1)} = +\frac{\lambda^2}{2}
$$

$$
E_1^{(2)} = \frac{|\langle 0|\lambda\hat{V}|1\rangle|^2}{E_1^{(0)} - E_0^{(0)}} = \frac{\lambda^2 \cdot 1}{-1 - 1} = -\frac{\lambda^2}{2}
$$

**The gap grows:**

$$
\Delta E = E_0 - E_1 = 2 + \lambda^2 + O(\lambda^4)
$$

The upper level rises, the lower level falls — they repel. This matches the exact result $2\sqrt{1+\lambda^2} = 2 + \lambda^2 - \frac{\lambda^4}{4} + \cdots$.

**General rule.** For any two levels with $E_n > E_m$, the second-order energy denominators satisfy $E_n - E_m > 0$ and $E_m - E_n < 0$, so:

$$
E_n^{(2)} = \sum_{k \neq n}\frac{|V_{kn}|^2}{E_n - E_k} > 0 \quad (\text{from the }k=m\text{ term}), \qquad E_m^{(2)} = \sum_{k \neq m}\frac{|V_{km}|^2}{E_m - E_k} < 0 \quad (\text{from the }k=n\text{ term})
$$

The coupling $V$ **always** pushes states of different energy apart at second order, regardless of the sign or magnitude of $V$. This is the perturbative origin of the universal level repulsion phenomenon.
:::


### Convergence

:::{admonition} Convergence Criterion
:class: important

The perturbative series converges when the perturbation is small compared to the energy gap:

$$

\lambda \|V\| \ll \Delta E

$$

For the qubit, the convergence radius is $R = 1$, set by the branch points at $\lambda = \pm\mathrm{i}$ in the complex plane where $E^2 = 1 + \lambda^2 = 0$.
:::

:::{admonition} Derivation: Convergence Radius from Branch Points
:class: dropdown information

The energy $E_\pm(\lambda) = \pm\sqrt{1+\lambda^2}$ is analytic in $\lambda$ except where $1+\lambda^2 = 0$, i.e., at $\lambda = \pm\mathrm{i}$. These are **branch points** of the square root.

The Taylor series centered at $\lambda = 0$ converges in the largest disk in the complex $\lambda$-plane that contains no singularity. The nearest singularity is at distance $|\lambda| = 1$ from the origin, so the convergence radius is

$$

R = \min_{\text{singularities}} |\lambda_*| = |\pm\mathrm{i}| = 1

$$

**Physical interpretation**: at $\lambda = \pm\mathrm{i}$ the two energy levels coalesce ($E_+ = E_- = 0$), forming an **exceptional point** — a non-Hermitian degeneracy (since $H(\pm\mathrm{i})$ is not Hermitian). The perturbative series, built on the assumption that the two levels are distinct, cannot be continued past this point.

More generally, for a $2\times 2$ system with gap $\Delta E$ and coupling $v$, the branch points are at $\lambda_* = \pm\mathrm{i}\,\Delta E/(2v)$, giving convergence radius $R = \Delta E/(2|v|)$, confirming the general criterion $\lambda|v| \ll \Delta E$.
:::

Beyond $|\lambda| = R$, the series diverges even though the exact solution remains valid. Perturbation theory breaks down precisely when the coupling becomes comparable to the energy gap — the perturbation can no longer be treated as "small."

:::{admonition} Discussion
:class: dropdown tip

The first-order corrections $E_\pm^{(1)} = 0$ vanish because $\hat{\sigma}^x$ has zero diagonal elements. This is not a coincidence but a consequence of parity symmetry. More generally, when does symmetry guarantee that certain perturbative corrections vanish? Can you think of physical examples where recognizing this saves enormous computational effort?
:::

### Summary

- **Perturbation theory** expands eigenvalues as $E_n(\lambda) = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \cdots$.
- The **qubit model** provides an exact benchmark: $E_\pm = \pm\sqrt{1+\lambda^2}$ reproduces all perturbative coefficients when Taylor-expanded.
- **Level repulsion**: coupled levels repel rather than cross; the gap is proportional to $|\langle 1|V|0\rangle|$.
- **Convergence**: the series converges for $\lambda\|V\| \ll \Delta E$; breakdown signals that the perturbation is no longer "small."
