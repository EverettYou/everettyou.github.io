# 2.1.1 Tensor Product

:::{admonition} Prompts
:class: tip

- What is a tensor product of two Hilbert spaces, and how do basis states combine? For a two-particle system where each particle has $d$ basis states, how many basis states does the combined system have?
- How do operators acting on individual particles extend to the full multi-particle Hilbert space? Why do operators on different particles always commute?
- What is a Pauli string, and why do Pauli strings form a complete basis for all multi-qubit Hermitian operators?
- How does the dimension of the $N$-particle Hilbert space grow with $N$? What does this exponential growth imply for classical simulation of quantum systems?
:::

## Lecture Notes

### Overview

Chapter 1 described a single qubit. To describe multiple quantum particles—atoms, photons, electrons in a solid—we need a systematic way to combine Hilbert spaces. The **tensor product** provides this: it constructs the full many-body state space from single-particle spaces, and its dimension grows exponentially with particle number. In this lesson we set up the many-body framework (states, operators, Pauli strings) that will be used throughout the rest of the course.

### Many-Body State Space

#### Tensor Product of Hilbert Spaces

:::{admonition} Tensor Product
:class: important

When systems A and B have Hilbert spaces $\mathcal{H}_A$ and $\mathcal{H}_B$, the composite system lives in

$$

\mathcal{H}_{AB} = \mathcal{H}_A \otimes \mathcal{H}_B

$$

If $\{\vert a_i\rangle\}$ is a basis for A and $\{\vert b_j\rangle\}$ a basis for B, then $\{\vert a_i\rangle \otimes \vert b_j\rangle\}$ is a complete orthonormal basis for $\mathcal{H}_{AB}$:

$$

\langle a_i b_j \vert a_{i'} b_{j'} \rangle = \langle a_i \vert a_{i'}\rangle \langle b_j \vert b_{j'}\rangle = \delta_{ii'}\delta_{jj'}

$$ (eq-tensor-inner-product)

:::

#### Two Qubits

Each qubit has basis $\{\vert 0\rangle, \vert 1\rangle\}$. The two-qubit space $\mathcal{H} = \mathbb{C}^2 \otimes \mathbb{C}^2$ has dimension $2 \times 2 = 4$, with basis:

$$

\vert 00\rangle,\quad \vert 01\rangle,\quad \vert 10\rangle,\quad \vert 11\rangle

$$

where $\vert ij\rangle \equiv \vert i\rangle_A \otimes \vert j\rangle_B$. Orthonormality: $\langle i_1 j_1 \vert i_2 j_2\rangle = \delta_{i_1 i_2}\delta_{j_1 j_2}$.

#### General Case: $N$ Particles

For $N$ particles, each with $d$ single-particle states $\vert\alpha\rangle$ ($\alpha = 1, \ldots, d$), the composite space is $\mathcal{H}^{\otimes N} = \mathcal{H}_1 \otimes \cdots \otimes \mathcal{H}_N$, with basis:

$$

\vert\alpha_1 \alpha_2 \cdots \alpha_N\rangle = \vert\alpha_1\rangle \otimes \vert\alpha_2\rangle \otimes \cdots \otimes \vert\alpha_N\rangle

$$

Here $\alpha$ is a **generic single-particle quantum number**: it can label momentum $\boldsymbol{k}$, position $\boldsymbol{r}$, spin $s$, an energy level, or any other set of quantum numbers that specifies a single-particle mode.

:::{admonition} Hilbert Space Dimension
:class: important

$$

\dim(\mathcal{H}^{\otimes N}) = d^N

$$ (eq-hilbert-dimension)

This grows **exponentially** with $N$. For $N = 300$ qubits: $2^{300} \approx 10^{90}$ dimensions—more than the number of atoms in the observable universe.
:::

A general state expands as:

$$

\vert\Psi\rangle = \sum_{\alpha_1, \ldots, \alpha_N} \Psi_{\alpha_1 \cdots \alpha_N}\;\vert\alpha_1 \cdots \alpha_N\rangle

$$ (eq-general-state)

with expansion coefficients satisfying $\sum \vert \Psi_{\alpha_1 \cdots \alpha_N}\vert^2 = 1$.


### Many-Body Operators

#### Single-Body Operators

An operator $\hat{A}$ acting on particle $i$ alone extends to the full space by tensoring with identity $\hat{I}$ on all other particles:

$$

\hat{A}_i = \hat{I}_1 \otimes \cdots \otimes \hat{A} \otimes \cdots \otimes \hat{I}_N \quad (\hat{A}\text{ in position }i)

$$ (eq-single-body-op)

Its action on basis states:

$$

\hat{A}_i \vert\alpha_1 \cdots \alpha_N\rangle = \vert\alpha_1\rangle \cdots (\hat{A}\vert\alpha_i\rangle) \cdots \vert\alpha_N\rangle

$$

:::{admonition} Operators on Different Particles Commute
:class: important

$$

[\hat{A}_i, \hat{B}_j] = 0 \quad \text{for } i \neq j

$$ (eq-different-particle-commute)

This is because they act on different tensor factors.
:::

:::{admonition} Derivation: Operators on Different Particles Commute
:class: dropdown information

Consider $N=2$ for clarity (the general case is identical). Let $\hat{A}_1 = \hat{A} \otimes \hat{I}$ and $\hat{B}_2 = \hat{I} \otimes \hat{B}$. Their products are:

$$

\hat{A}_1 \hat{B}_2 = (\hat{A} \otimes \hat{I})(\hat{I} \otimes \hat{B}) = \hat{A}\hat{I} \otimes \hat{I}\hat{B} = \hat{A} \otimes \hat{B}

$$

$$

\hat{B}_2 \hat{A}_1 = (\hat{I} \otimes \hat{B})(\hat{A} \otimes \hat{I}) = \hat{I}\hat{A} \otimes \hat{B}\hat{I} = \hat{A} \otimes \hat{B}

$$

where we used the tensor product multiplication rule $(\hat{C}_1 \otimes \hat{C}_2)(\hat{D}_1 \otimes \hat{D}_2) = \hat{C}_1\hat{D}_1 \otimes \hat{C}_2\hat{D}_2$. Both products give the same result, so $[\hat{A}_1, \hat{B}_2] = 0$.

For general $N$: all tensor factors except $i$ and $j$ carry $\hat{I}$ in both operators, so they multiply trivially. The factors at positions $i$ and $j$ involve $\hat{A} \cdot \hat{I}$ vs $\hat{I} \cdot \hat{A}$ and $\hat{I} \cdot \hat{B}$ vs $\hat{B} \cdot \hat{I}$, which commute.
:::

#### Two-Body Operators

An operator coupling particles $i$ and $j$ acts nontrivially on both tensor factors:

$$

\hat{V}_{ij} = \hat{I}_1 \otimes \cdots \otimes \hat{V} \otimes \cdots \otimes \hat{I}_N \quad (\hat{V}\text{ acts on positions }i,j)

$$

:::{admonition} Example: Heisenberg Spin Coupling
:class: dropdown example

**Problem.** Two spin-1/2 particles interact via $\hat{H}_{\mathrm{int}} = J\,\hat{\boldsymbol{S}}_1 \cdot \hat{\boldsymbol{S}}_2$.

**Solution.** Expanding in components:

$$

\hat{H}_{\mathrm{int}} = J(\hat{S}_1^x \hat{S}_2^x + \hat{S}_1^y \hat{S}_2^y + \hat{S}_1^z \hat{S}_2^z)

$$

where each $\hat{S}_i^a$ acts on particle $i$ only. The sign of $J$ determines whether alignment is favored ($J < 0$, ferromagnetic) or disfavored ($J > 0$, antiferromagnetic).
:::

#### General Operator Expansion

Any operator on the $N$-particle space can be expanded in the basis of ket-bra products:

$$

\hat{O} = \sum_{\{\alpha\},\{\beta\}} O_{\alpha_1\cdots\alpha_N,\,\beta_1\cdots\beta_N}\;\vert\alpha_1\cdots\alpha_N\rangle\langle\beta_1\cdots\beta_N\vert

$$

For multi-qubit systems, a more natural basis is given by Pauli strings.

#### Pauli Strings

:::{admonition} Pauli String Expansion
:class: important

Any Hermitian operator on $N$ qubits can be written as:

$$

\hat{O} = \sum_{\{s_i\}} c_{\{s_i\}} \bigotimes_{i=1}^N \hat{\sigma}_i^{s_i}, \qquad \hat{\sigma}_i^{s_i} \in \{\hat{I},\, \hat{X},\, \hat{Y},\, \hat{Z}\}

$$ (eq-pauli-string)

with $c_{\{s_i\}} \in \mathbb{R}$. There are $4^N$ Pauli strings; they form an orthogonal basis for all $2^N \times 2^N$ Hermitian matrices.
:::

Each Pauli string $\bigotimes_i \hat{\sigma}_i^{s_i}$ is itself a tensor product of single-qubit operators. The decomposition is unique because the $4^N$ Pauli strings are orthogonal under the trace inner product $\langle \hat{A}, \hat{B}\rangle = \mathrm{Tr}(\hat{A}^\dagger \hat{B})$.

:::{admonition} Derivation: Orthogonality of Pauli Strings
:class: dropdown information

**Single-qubit case.** The four single-qubit Pauli matrices $\{\hat{I}, \hat{X}, \hat{Y}, \hat{Z}\}$ satisfy:

$$

\mathrm{Tr}(\hat{\sigma}^a \hat{\sigma}^b) = 2\,\delta_{ab}

$$

for $a, b \in \{I, X, Y, Z\}$. This follows from $\hat{\sigma}^a \hat{\sigma}^b = \delta_{ab}\hat{I} + \mathrm{i}\epsilon_{abc}\hat{\sigma}^c$ (for $a,b \neq I$), together with $\mathrm{Tr}(\hat{I}) = 2$ and $\mathrm{Tr}(\hat{\sigma}^c) = 0$.

**Multi-qubit case.** For two Pauli strings $\hat{P} = \bigotimes_{i=1}^N \hat{\sigma}_i^{a_i}$ and $\hat{Q} = \bigotimes_{i=1}^N \hat{\sigma}_i^{b_i}$, the trace factorizes:

$$

\mathrm{Tr}(\hat{P}^\dagger \hat{Q}) = \prod_{i=1}^N \mathrm{Tr}(\hat{\sigma}_i^{a_i} \hat{\sigma}_i^{b_i}) = \prod_{i=1}^N 2\,\delta_{a_i b_i} = 2^N \prod_{i=1}^N \delta_{a_i b_i}

$$

This equals $2^N$ if the two strings are identical (all $a_i = b_i$) and zero otherwise. Since there are $4^N$ orthogonal strings and the space of $2^N \times 2^N$ Hermitian matrices is also $4^N$-dimensional, the Pauli strings form a complete orthogonal basis. The expansion coefficients are extracted by:

$$

c_{\{s_i\}} = \frac{1}{2^N}\,\mathrm{Tr}\!\Bigl(\hat{O}\,\bigotimes_{i=1}^N \hat{\sigma}_i^{s_i}\Bigr)

$$

:::

:::{admonition} Example: Pauli String Decomposition
:class: dropdown example

**Problem.** Decompose the projector $\hat{P} = \vert 00\rangle\langle 00\vert$ into Pauli strings.

**Solution.** Use $\vert 0\rangle\langle 0\vert = \tfrac{1}{2}(\hat{I} + \hat{Z})$:

$$

\vert 00\rangle\langle 00\vert = \vert 0\rangle\langle 0\vert \otimes \vert 0\rangle\langle 0\vert = \tfrac{1}{4}(\hat{I} + \hat{Z}) \otimes (\hat{I} + \hat{Z}) = \tfrac{1}{4}(\hat{I}\hat{I} + \hat{I}\hat{Z} + \hat{Z}\hat{I} + \hat{Z}\hat{Z})

$$

Four Pauli strings, each with real coefficient $\tfrac{1}{4}$.
:::

:::{admonition} Discussion: Exponential Complexity
:class: dropdown tip

The number of Pauli strings grows as $4^N$, which is exponential in $N$. This means specifying a general $N$-qubit Hamiltonian requires exponentially many parameters. In practice, physical Hamiltonians are **local** (each term couples only a few nearby qubits), so only polynomially many Pauli strings have nonzero coefficients. What are the implications for classical simulation of quantum systems?
:::



:::{admonition} Derivation: Entanglement Exists — The Bell State Cannot Be Factored
:class: dropdown information

**Claim.** The two-qubit state $\vert\Phi^+\rangle = \tfrac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$ is **entangled**: it cannot be written as a product state $\vert a\rangle \otimes \vert b\rangle$ for any single-qubit states $\vert a\rangle, \vert b\rangle$.

**Proof by contradiction.** Suppose

$$
\vert\Phi^+\rangle = (\alpha\vert 0\rangle + \beta\vert 1\rangle) \otimes (\gamma\vert 0\rangle + \delta\vert 1\rangle)
= \alpha\gamma\vert 00\rangle + \alpha\delta\vert 01\rangle + \beta\gamma\vert 10\rangle + \beta\delta\vert 11\rangle
$$

for some $\alpha, \beta, \gamma, \delta \in \mathbb{C}$. Matching the four basis-state coefficients of $\tfrac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$:

| Basis state | $\vert\Phi^+\rangle$ coefficient | Product-state coefficient |
|-------------|---------------------|--------------------------|
| $\vert 00\rangle$ | $1/\sqrt{2}$ | $\alpha\gamma$ |
| $\vert 01\rangle$ | $0$ | $\alpha\delta$ |
| $\vert 10\rangle$ | $0$ | $\beta\gamma$ |
| $\vert 11\rangle$ | $1/\sqrt{2}$ | $\beta\delta$ |

From the $\vert 01\rangle$ equation: $\alpha\delta = 0$, so either $\alpha = 0$ or $\delta = 0$.

- **Case 1: $\alpha = 0$.** Then $\alpha\gamma = 0 \neq 1/\sqrt{2}$. Contradiction.
- **Case 2: $\delta = 0$.** Then $\beta\delta = 0 \neq 1/\sqrt{2}$. Contradiction.

Both cases are impossible. Therefore no product-state factorization exists. $\blacksquare$

**Why this matters.** The fact that entangled states exist shows that the tensor product of Hilbert spaces is strictly larger than the set of product states. Most states in $\mathcal{H}_A \otimes \mathcal{H}_B$ are entangled — the set of separable (product) states has measure zero in the full state space. The four Bell states $\vert\Phi^\pm\rangle = (\vert 00\rangle \pm \vert 11\rangle)/\sqrt{2}$ and $\vert\Psi^\pm\rangle = (\vert 01\rangle \pm \vert 10\rangle)/\sqrt{2}$ form an orthonormal basis of the two-qubit space consisting entirely of entangled states.
:::

:::{admonition} Derivation: The Swap Operator in the Pauli String Basis
:class: dropdown information

The **swap operator** $\hat{S}$ exchanges two particles:

$$
\hat{S}\vert ab\rangle = \vert ba\rangle \quad \text{for all basis states } a, b\in\{0,1\}
$$

We show it equals a particular Pauli string sum, then deduce its spectrum.

**Claim.**

$$
\hat{S} = \frac{1}{2}\bigl(\hat{I}\otimes\hat{I} + \hat{\sigma}^x\otimes\hat{\sigma}^x + \hat{\sigma}^y\otimes\hat{\sigma}^y + \hat{\sigma}^z\otimes\hat{\sigma}^z\bigr) = \frac{1}{2}\sum_{\mu\in\{0,x,y,z\}}\hat{\sigma}^\mu\otimes\hat{\sigma}^\mu
$$

**Proof via the Pauli completeness relation.** The four matrices $\{\hat{I},\hat{\sigma}^x,\hat{\sigma}^y,\hat{\sigma}^z\}$ form a complete (Hilbert–Schmidt-orthonormal) basis for all $2\times 2$ operators (see §1.1.3). Any $2\times 2$ matrix $M$ expands as:

$$
M = \sum_\mu \frac{\mathrm{Tr}(\hat{\sigma}^\mu M)}{2}\,\hat{\sigma}^\mu
$$

Apply this to the rank-1 "matrix unit" $M = \vert a\rangle\langle c\vert$:

$$
\vert a\rangle\langle c\vert = \sum_\mu \frac{\mathrm{Tr}(\hat{\sigma}^\mu \vert a\rangle\langle c\vert)}{2}\,\hat{\sigma}^\mu = \sum_\mu \frac{(\hat{\sigma}^\mu)_{ca}}{2}\,\hat{\sigma}^\mu
$$

Taking matrix elements $\langle b\vert\cdot\vert d\rangle$ of both sides:

$$
\delta_{ab}\delta_{cd} = \sum_\mu \frac{(\hat{\sigma}^\mu)_{ca}}{2}\,(\hat{\sigma}^\mu)_{bd} \quad \Longrightarrow \quad \sum_\mu (\hat{\sigma}^\mu)_{ac}(\hat{\sigma}^\mu)_{bd} = 2\delta_{ad}\delta_{bc}
$$

The matrix element of the candidate operator $\tfrac{1}{2}\sum_\mu \hat{\sigma}^\mu\otimes\hat{\sigma}^\mu$ in the two-qubit basis $\vert ab\rangle$ is:

$$
\Bigl\langle ab\Bigl\vert\frac{1}{2}\sum_\mu\hat{\sigma}^\mu\otimes\hat{\sigma}^\mu\Bigr\vert cd\Bigr\rangle = \frac{1}{2}\sum_\mu(\hat{\sigma}^\mu)_{ac}(\hat{\sigma}^\mu)_{bd} = \delta_{ad}\delta_{bc}
$$

The matrix element of $\hat{S}$ is $\langle ab\vert\hat{S}\vert cd\rangle = \langle ab\vert dc\rangle = \delta_{ad}\delta_{bc}$.

The two matrix elements are identical, so $\hat{S} = \tfrac{1}{2}\sum_\mu\hat{\sigma}^\mu\otimes\hat{\sigma}^\mu$. $\blacksquare$

**Spectrum.** Since $\hat{S}^2 = \hat{I}$ (swapping twice returns to the original state), the eigenvalues of $\hat{S}$ are $\pm 1$:

- **Symmetric subspace** ($\hat{S} = +1$): the three triplet states $\{\vert 00\rangle,\,\tfrac{1}{\sqrt{2}}(\vert 01\rangle+\vert 10\rangle),\,\vert 11\rangle\}$.
- **Antisymmetric subspace** ($\hat{S} = -1$): the singlet $\vert\Psi^-\rangle = \tfrac{1}{\sqrt{2}}(\vert 01\rangle-\vert 10\rangle)$.

From the Pauli string form: $\hat{S} = \tfrac{1}{2}\hat{I}\otimes\hat{I} + \tfrac{1}{2}\hat{\boldsymbol{\sigma}}_1\cdot\hat{\boldsymbol{\sigma}}_2$, so the eigenvalues of $\hat{\boldsymbol{\sigma}}_1\cdot\hat{\boldsymbol{\sigma}}_2$ (the Heisenberg exchange interaction) are $+1$ (triplet) and $-3$ (singlet). This is the foundation of the Heisenberg spin model in condensed matter physics and directly anticipates the symmetrization postulate of §2.1.2.
:::

:::{admonition} Derivation: Schmidt Decomposition via Singular Value Decomposition
:class: dropdown information

Every pure state $\vert\psi\rangle \in \mathcal{H}_A \otimes \mathcal{H}_B$ (with $\dim\mathcal{H}_A = d_A$, $\dim\mathcal{H}_B = d_B$) can be written as:

$$
\vert\psi\rangle = \sum_{k=1}^{r} \lambda_k \vert u_k\rangle_A \vert v_k\rangle_B
$$

where $\{\vert u_k\rangle\}$ and $\{\vert v_k\rangle\}$ are orthonormal sets, $\lambda_k > 0$, and $r \leq \min(d_A, d_B)$ is the **Schmidt rank**.

**Step 1: Coefficient matrix.** Choose fixed orthonormal bases $\{\vert i\rangle_A\}$, $\{\vert j\rangle_B\}$. Expand:

$$
\vert\psi\rangle = \sum_{i,j} C_{ij} \vert i\rangle_A \vert j\rangle_B
$$

where the coefficient matrix $C$ is $d_A \times d_B$ with $\sum_{ij}\vert C_{ij}\vert^2 = 1$ (from normalization).

**Step 2: Apply SVD.** By the singular value decomposition, any complex matrix $C$ can be written as $C = U \Lambda V^\dagger$, where:
- $U$ is $d_A \times d_A$ unitary with columns $\boldsymbol{u}_k$
- $V$ is $d_B \times d_B$ unitary with columns $\boldsymbol{v}_k$
- $\Lambda = \mathrm{diag}(\lambda_1, \ldots, \lambda_r, 0, \ldots)$ with $\lambda_k > 0$ (singular values)

So $C_{ij} = \sum_k U_{ik} \lambda_k V_{jk}^*$.

**Step 3: Substitute back.**

$$
\vert\psi\rangle = \sum_{i,j} \sum_k U_{ik}\lambda_k V_{jk}^* \vert i\rangle_A \vert j\rangle_B
= \sum_k \lambda_k \underbrace{\Bigl(\sum_i U_{ik}\vert i\rangle_A\Bigr)}_{= \vert u_k\rangle_A} \underbrace{\Bigl(\sum_j V_{jk}^*\vert j\rangle_B\Bigr)}_{= \vert v_k\rangle_B}
$$

The new states $\vert u_k\rangle_A = \sum_i U_{ik}\vert i\rangle_A$ and $\vert v_k\rangle_B = \sum_j V_{jk}^*\vert j\rangle_B$ form orthonormal sets because $U$ and $V$ are unitary: $\langle u_k\vert u_{k'}\rangle = \sum_i U_{ik}^* U_{ik'} = (U^\dagger U)_{kk'} = \delta_{kk'}$.

**Step 4: Normalization.** $\sum_k \lambda_k^2 = \sum_k (\Lambda^\dagger\Lambda)_{kk} = \mathrm{Tr}(V C^\dagger U^\dagger U C V^\dagger) = \mathrm{Tr}(C^\dagger C) = \sum_{ij}\vert C_{ij}\vert^2 = 1$.

**Step 5: Entanglement criterion.** The reduced density matrix of $A$ is:

$$
\hat{\rho}_A = \mathrm{Tr}_B(\vert\psi\rangle\langle\psi\vert) = \sum_k \lambda_k^2 \vert u_k\rangle\langle u_k\vert
$$

- **Schmidt rank 1** ($r = 1$): $\vert\psi\rangle = \vert u_1\rangle_A\vert v_1\rangle_B$ is a **product state**; $\hat{\rho}_A = \vert u_1\rangle\langle u_1\vert$ is pure.
- **Schmidt rank** $r \geq 2$: $\hat{\rho}_A$ has at least two nonzero eigenvalues — the state is **entangled**; $\mathrm{Tr}(\hat{\rho}_A^2) = \sum_k \lambda_k^4 < 1$.

The Schmidt rank is thus the cleanest basis-independent criterion for bipartite entanglement. $\square$
:::

### Summary

- **Tensor product** $\mathcal{H}^{\otimes N}$ combines single-particle spaces; dimension grows as $d^N$.
- **Basis states** $\vert\alpha_1 \cdots \alpha_N\rangle$ are products of single-particle states; they form a complete orthonormal set.
- **Single-body operators** extend via $\hat{I}$ on other particles; operators on different particles commute.
- **Two-body operators** couple pairs of particles, creating correlations.
- **Pauli strings** $\bigotimes_i \hat{\sigma}_i^{s_i}$ form an orthogonal basis for all multi-qubit Hermitian operators, with real coefficients.
