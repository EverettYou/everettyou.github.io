# 2.2 Angular Momentum

## Overview

Angular momentum is one of the most fundamental concepts in quantum mechanics. It governs atomic structure, chemical bonding, magnetism, and even connects to relativistic physics through the spin-statistics theorem.

**Key applications:**

1. **Atomic structure**: Electrons fill orbitals respecting spin-pairing rules (two electrons per orbital, opposite spins).
2. **Chemistry**: Chemical bonding depends on electron spin coupling.
3. **Magnetism**: Unpaired electron spins create magnetic moments. Ferromagnetism is collective alignment of spins.
4. **Particle physics**: The spin-statistics theorem connects intrinsic angular momentum to permutation symmetry—a profound consequence of relativistic quantum field theory.
5. **Quantum computing**: Spin qubits (in semiconductors) and spin-1/2 systems are leading platforms for quantum information.
6. **Precision measurements**: Atomic clocks rely on spin properties; GPS and modern navigation depend on them.

### Fundamental Relations

All angular momenta (orbital, spin, total) satisfy:

$$


[J_i, J_j] = \mathrm{i}\hbar \epsilon_{ijk} J_k

$$


**Eigenvalues** of simultaneous eigenstates $|j, m\rangle$:

$$


J^2 |j, m\rangle = \hbar^2 j(j+1) |j, m\rangle, \quad J_z |j, m\rangle = \hbar m |j, m\rangle

$$


where:
- $j \in \{0, 1/2, 1, 3/2, \ldots\}$ (angular momentum quantum number)
- $m \in \{-j, -j+1, \ldots, j-1, j\}$ (magnetic quantum number)

### Spin-1/2 (The Simplest Case)

For **spin-1/2 particles** (electrons, quarks, neutrinos):

- **Basis states**: $|\uparrow\rangle$ (spin up) and $|\downarrow\rangle$ (spin down)
- **Operators**: $S_i = \frac{\hbar}{2} \hat{\sigma}_i$ where $\hat{\sigma}_i$ are **Pauli matrices**
- **Unique feature**: Spinor transformation picks up phase $-1$ under $2\pi$ rotation; returns to original only after $4\pi$ rotation
- **Total spin squared**: $S^2 = \frac{3\hbar^2}{4}$

**Two-spin combinations:**

- **Singlet state** ($j=0$, antisymmetric): $|0, 0\rangle = \frac{1}{\sqrt{2}}(|\uparrow\downarrow\rangle - |\downarrow\uparrow\rangle)$ (opposite spins)
- **Triplet states** ($j=1$, symmetric): $|1, m\rangle$ for $m = -1, 0, +1$ (same or opposite spins)

### Spin-Statistics Connection

| Spin | Statistics | Examples |
|------|-----------|----------|
| Integer ($s = 0, 1, 2, \ldots$) | Boson | Photons, pions, helium-4 |
| Half-integer ($s = 1/2, 3/2, \ldots$) | Fermion | Electrons, quarks, neutrinos |


### Key Concepts

- **Angular momentum algebra**: $[J_i, J_j] = \mathrm{i}\hbar \epsilon_{ijk} J_k$ governs all angular momentum in quantum mechanics.
- **Casimir operator**: $J^2$ commutes with all components; simultaneous eigenstates $|j, m\rangle$ exist.
- **Spin-1/2 representation**: Pauli matrices $\hat{\sigma}^x, \hat{\sigma}^y, \hat{\sigma}^z$ as generators for the simplest nontrivial representation.
- **Addition of angular momentum**: Combining two spins via Clebsch-Gordan coefficients; singlet and triplet states arise from coupling.
- **Spin-orbit coupling**: Interaction between orbital and spin angular momentum, producing fine structure splitting.
- **Spin-statistics theorem**: Spin and exchange statistics are inextricably linked by relativistic quantum field theory.

### Learning Objectives

- Derive the commutation relations for angular momentum and apply them to construct the spin-1/2 representation using Pauli matrices.
- Compute eigenvalues and eigenstates of $J^2$ and $J_z$ for single and composite spins.
- Construct singlet and triplet states from two spin-1/2 particles using Clebsch-Gordan coefficients.
- Apply ladder operators to systematically generate all eigenstates from a single reference state.
- Explain the spin-statistics theorem and predict exchange symmetry from intrinsic angular momentum.
- Understand spin-orbit coupling and its role in atomic fine structure.


## Projects

### Project: Ultracold Fermi Gases and the BCS-BEC Crossover

**Objective:** Model the BCS-BEC crossover in ultracold Fermi gases—the continuous evolution from weakly-paired fermions (Cooper pairs) to tightly-bound bosonic molecules as interactions are tuned.

**Background:** The BCS-BEC crossover is a frontier research area in cold atoms and condensed matter. Using Feshbach resonances, experimentalists can tune the scattering length $a$ continuously and observe the transformation from BCS superconductor (large $N$, small gap) to BEC of tightly-bound molecules. Angular momentum plays a crucial role: identical fermions require odd relative angular momentum ($\ell = 1, 3, \ldots$), constraining pairing. This project deepens understanding of identical particle symmetry (Chapter 2) by studying competing quantum phases and the role of angular momentum in many-body physics.

**Suggested Approach:**
- **Literature survey**: Read "The BCS-BEC crossover and the unitary Fermi gas" (Giorgini, Pitaevskii & Stringari 2008, Rev. Mod. Phys.). Understand the mean-field equations in the BCS and BEC limits.
- **Theory**: Derive the gap equation and chemical potential in mean-field BCS theory. Show how the ground state energy, condensate fraction, and excitation gap depend on interaction strength $a$.
- **Computational method**: Solve the BCS equations numerically using self-consistent iteration. Parametrize by the scattering length or dimensionless coupling $(k_F a)^{-1}$, where $k_F$ is the Fermi wavenumber.
- **Phase diagram**: Plot the gap $\Delta$, condensate fraction, and coherence length $\xi = \hbar^2/(m\Delta)$ across the crossover. Show that at unitarity ($(k_F a) = \infty$), universal behavior emerges.
- **Excitations**: Study fermionic excitations (Bogoliubov spectrum) vs. bosonic excitations (collective modes). Show the transition from weak-coupling fermions to strong-coupling molecules.
- **Angular momentum**: Analyze why $\ell = 1$ pairing dominates for spin-up/spin-down fermions. Include effects of angular momentum quantization on pairing stability.

**Expected Deliverable:** Research report (7-10 pages) with: BCS-BEC crossover theory, numerical solution of gap equation, phase diagram and universal properties at unitarity, excitation spectra analysis, comparison with experimental data from ultracold Fermi gas experiments, and physical interpretation connecting angular momentum constraints to observable phenomena.

**Key References:** Giorgini, Pitaevskii & Stringari (2008) "Theory of ultracold atomic Fermi gases"; Ketterle & Zwierlein "Making, probing and understanding ultracold Fermi gases" (2008); experimental papers from JILA, MIT, or LENS labs on Fermi gas crossover.
