## Lecture Notes

### Overview

The fractional quantum Hall effect (FQHE) is the premier experimental realization of anyons: at filling fraction $
u = 1/m$, the ground state is an incompressible quantum liquid whose elementary excitations carry charge $e/m$ and anyonic exchange statistics $	heta = \pi/m$. This lesson applies the charge-flux composite picture to explain both the fractional charge and the fractional statistics from a single Aharonov-Bohm phase calculation.

### The Fractional Quantum Hall Effect

When a 2D electron gas is placed in a strong perpendicular magnetic field at low temperature, the Hall conductance is quantized. At certain fractional filling factors, the system forms an incompressible quantum liquid:

:::{admonition} Filling Fractions and the Hierarchy
:class: important

The fractional quantum Hall effect (FQHE) occurs at filling fractions $\nu = 1/3, 2/5, 1/5, 2/3, \ldots$ The most prominent is $\nu = 1/3$, discovered by Tsui, Störmer, and Gossard in 1982 (Nobel Prize 1998).
:::

The FQHE cannot be explained by single-particle physics — it arises from strong electron-electron correlations. The ground state is described by the **Laughlin wavefunction**:

$$\psi(z_1, \ldots, z_N) = \prod_{i < j} (z_i - z_j)^m \, \mathrm{e}^{-\sum_k |z_k|^2 / 4\ell_B^2}$$

where $z_k = x_k + \mathrm{i} y_k$ are complex coordinates, $m$ is an odd integer, $\ell_B = \sqrt{\hbar/eB}$ is the magnetic length, and $\nu = 1/m$.
:::{admonition} Derivation: Filling Fraction $\nu = 1/m$ from the Laughlin Wavefunction
:class: dropdown information

The Laughlin wavefunction for $N$ electrons is

$$
\psi(z_1, \ldots, z_N) = \prod_{i < j} (z_i - z_j)^m \, \mathrm{e}^{-\sum_k |z_k|^2 / 4\ell_B^2}
$$

where $z_k = x_k + \mathrm{i} y_k$ and the magnetic length is $\ell_B = \sqrt{\hbar/eB}$.

**Step 1: Highest power of any single coordinate.** Focus on $z_1$. It appears in $(z_1 - z_j)^m$ for each $j = 2, \ldots, N$, so the highest power of $z_1$ in the polynomial part is $m(N-1)$.

**Step 2: Relate to Landau level orbitals.** In the lowest Landau level (LLL), the single-particle orbitals are $\phi_n(z) \propto z^n \, \mathrm{e}^{-|z|^2/4\ell_B^2}$ for $n = 0, 1, 2, \ldots$ The orbital $\phi_n$ is localized on a ring of radius $r_n = \sqrt{2n}\,\ell_B$. Since the highest power is $m(N-1)$, the wavefunction uses orbitals up to $n_{\max} = m(N-1)$.

**Step 3: Count flux quanta.** The number of available single-particle states (flux quanta threading the sample) is $N_\phi = n_{\max} + 1 \approx m(N-1) + 1 \approx mN$ for large $N$.

**Step 4: Filling fraction.** The filling fraction is the ratio of electrons to flux quanta:

$$
\nu = \frac{N}{N_\phi} = \frac{N}{mN} = \frac{1}{m}
$$

For $m = 3$: each electron occupies $m = 3$ flux quanta worth of space, giving $\nu = 1/3$.
:::


:::{admonition} Anyon Excitation Properties at $\nu = 1/3$
:class: important

The elementary excitations of the $\nu = 1/3$ state are **anyons** with:

- **Fractional charge**: $q = e/3$ (one-third of the electron charge)
- **Fractional statistics**: statistical angle $\theta = \pi/3$ (exchange phase $\mathrm{e}^{\mathrm{i}\pi/3}$)

These properties have been confirmed experimentally through shot noise measurements (charge) and interferometry (statistics).
:::
:::{admonition} Derivation: Fractional Charge $e/m$ via Laughlin's Gauge Argument
:class: dropdown information

Laughlin's original argument (1983) derives the quasihole charge by adiabatically inserting a flux quantum through the center of a disk-shaped sample.

**Setup.** Consider $N$ electrons in the $\nu = 1/m$ Laughlin state on a disk. Thread a thin solenoid through the center, and slowly increase the flux $\Phi$ from $0$ to $\Phi_0 = h/e$ (one flux quantum).

**Step 1: Gauge invariance.** By the Aharonov-Bohm effect, inserting flux $\Phi$ into the solenoid is equivalent to the gauge transformation $z_k \to z_k$ with each orbital shifting: $\phi_n(z) \propto z^n \to z^{n+1}$. After inserting exactly one flux quantum $\Phi_0$, each single-particle orbital maps to the next: $n \to n + 1$.

**Step 2: Effect on the Laughlin state.** Under $z_k^n \to z_k^{n+1}$ for all $k$, the Laughlin wavefunction transforms as

$$
\prod_{i<j} (z_i - z_j)^m \to \left(\prod_k z_k\right) \prod_{i<j}(z_i - z_j)^m
$$

The new state has a **quasihole** at the origin: the factor $\prod_k z_k$ forces all electrons away from $z = 0$.

**Step 3: Charge of the quasihole.** The Hall conductance at $\nu = 1/m$ is $\sigma_{xy} = \nu \, e^2/h = e^2/(mh)$. The radial current response to the adiabatic flux insertion transfers charge

$$
Q = \sigma_{xy} \cdot \Phi_0 = \frac{e^2}{mh} \cdot \frac{h}{e} = \frac{e}{m}
$$

from the interior to the edge. The quasihole left behind therefore carries charge $q^* = e/m$.

For $m = 3$: inserting one flux quantum creates a quasihole of charge $e/3$.
:::


### Applying the Charge-Flux Picture

The charge-flux composite picture developed in §2.3.1 explains why FQHE anyon excitations have fractional statistics. At $\nu = 1/m$, each electron binds $m$ flux quanta, and the anyon excitation carries charge $q^* = e/m$ and one flux quantum $\Phi^* = \Phi_0 = h/e$.

:::{admonition} Example: $\nu = 1/3$ Anyons
:class: note

An anyon excitation at $\nu = 1/3$ has charge $q^* = e/3$ and carries one flux quantum $\Phi^* = \Phi_0 = h/e$. Applying the braiding phase {eq}`eq-braid-phase`:

$$

\theta_{\mathrm{br}} = \frac{q^*\,\Phi^*}{\hbar} = \frac{(e/3)(h/e)}{\hbar} = \frac{2\pi}{3}

$$

The exchange phase is half the braiding phase: $\theta_{\mathrm{ex}} = \theta_{\mathrm{br}}/2 = \pi/3$. The spin quantum number is $s = \theta_{\mathrm{ex}}/(2\pi) = 1/6$.
:::

:::{admonition} Derivation: Anyon Statistics at General $\nu = 1/m$
:class: dropdown information

At filling $\nu = 1/m$ (with $m$ an odd integer), the quasihole has charge $q^* = e/m$ and carries one flux quantum $\Phi^* = \Phi_0 = h/e$. Using the charge-flux composite picture from §2.3.1:

**Braiding phase** (one quasihole makes a full loop around another):

$$
\theta_{\mathrm{br}} = \frac{q^* \Phi^*}{\hbar} = \frac{(e/m)(h/e)}{\hbar} = \frac{h}{m\hbar} = \frac{2\pi}{m}
$$

**Exchange phase** (half-braid):

$$
\theta_{\mathrm{ex}} = \frac{\theta_{\mathrm{br}}}{2} = \frac{\pi}{m}
$$

**Topological spin:**

$$
s = \frac{\theta_{\mathrm{ex}}}{2\pi} = \frac{1}{2m}
$$

**Verification against the Laughlin wavefunction.** The same result follows directly from the analytic structure of $\psi$. When two quasiholes at positions $\eta_1, \eta_2$ are exchanged, the wavefunction acquires a phase from the factor

$$
(\eta_1 - \eta_2)^{1/m}
$$

that appears in the quasihole-quasihole correlation. Under the exchange $\eta_1 \leftrightarrow \eta_2$, this factor picks up $\mathrm{e}^{\mathrm{i}\pi/m}$, confirming $\theta_{\mathrm{ex}} = \pi/m$.

| $\nu = 1/m$ | Charge $q^*$ | $\theta_{\mathrm{br}}$ | $\theta_{\mathrm{ex}}$ | Spin $s$ |
|---|---|---|---|---|
| $1/3$ | $e/3$ | $2\pi/3$ | $\pi/3$ | $1/6$ |
| $1/5$ | $e/5$ | $2\pi/5$ | $\pi/5$ | $1/10$ |
| $1/7$ | $e/7$ | $2\pi/7$ | $\pi/7$ | $1/14$ |
:::

The charge-flux picture reduces the exotic phenomenon of fractional statistics to the Aharonov-Bohm effect — a charge moving around a flux picks up a phase proportional to the enclosed flux.

:::{admonition} Derivation: Laughlin's Plasma Analogy and Ground State Density
:class: dropdown information

The Laughlin wavefunction can be analyzed exactly using an analogy to a classical 2D plasma, revealing why the electron density is uniform at $\rho_0 = \nu/(2\pi\ell_B^2)$.

**Rewrite as a Boltzmann weight.** Write $|\psi|^2 = \mathrm{e}^{-\beta U_\mathrm{cl}}$ where we identify $\beta = 2/m$ and the classical energy:

$$
U_\mathrm{cl} = -m^2 \sum_{i<j} \ln|z_i - z_j| + \frac{m}{2}\sum_k \frac{|z_k|^2}{2\ell_B^2}
$$

This is the energy of a **2D one-component plasma** (OCP): $N$ "charges" of magnitude $m$ embedded in a uniform neutralizing background with charge density $\rho_\mathrm{bg} = 1/(2\pi\ell_B^2)$.

**Plasma equilibrium.** In the OCP, the particles distribute to screen the background charge. The equilibrium condition (Poisson equation in 2D) requires that each plasma particle sits in a potential created by all others plus the background. The system reaches electrical neutrality when the particle density satisfies

$$
m \cdot \rho_\mathrm{el} = \rho_\mathrm{bg} = \frac{1}{2\pi\ell_B^2}
$$

Solving: $\rho_\mathrm{el} = 1/(2\pi m \ell_B^2)$.

**Electron density and filling fraction.** Since each plasma particle represents one electron, the electron number density is

$$
\rho_e = \frac{1}{2\pi m \ell_B^2}
$$

The number of flux quanta per unit area is $1/(2\pi\ell_B^2)$ (one flux quantum per area $2\pi\ell_B^2$). The filling fraction is therefore

$$
\nu = \frac{\rho_e}{1/(2\pi\ell_B^2)} = \frac{1}{m}
$$

confirming that the Laughlin state describes a uniform, incompressible liquid at filling $\nu = 1/m$. The plasma analogy provides the key insight: the Jastrow factor $\prod_{i<j}|z_i-z_j|^m$ is not just an ansatz — it encodes the optimal screening of the background potential.
:::



### Emergent Statistics

:::{admonition} Statistics Is Emergent
:class: attention

The anyonic statistics of FQHE quasiparticles is **not** a fundamental property of elementary particles — it **emerges** from the collective behavior of many strongly correlated electrons in a magnetic field. A single electron in vacuum is always a fermion. Only when $\sim 10^{10}$ electrons form an incompressible quantum liquid does the effective low-energy description produce quasiparticles with fractional charge and fractional statistics. This is a hallmark of **topological order**: the statistics is encoded in the global entanglement pattern of the ground state, not in any local property.
:::


:::{admonition} Derivation: Composite Fermion Picture and the Jain Sequence
:class: dropdown information

The **composite fermion** (CF) construction of Jain extends the Laughlin description to a broader family of filling fractions $\nu = p/(2sp + 1)$ for integers $p, s \geq 1$.

**Step 1: Attach flux quanta.** Attach $2s$ flux quanta $\Phi_0 = h/e$ to each electron to form a **composite fermion**. The attachment is achieved via a singular gauge transformation (Chern-Simons transformation); the resulting quasi-particles are still fermions (attaching an even number of flux quanta preserves statistics).

**Step 2: Mean-field effective magnetic field.** In a uniform electron density $n_e$, the attached flux partially cancels the external field:

$$
B^* = B - 2s n_e \Phi_0.
$$

At filling fraction $\nu = n_e h / (eB)$, substituting $n_e = \nu eB/h$:

$$
B^* = B\left(1 - 2s\nu\right).
$$

**Step 3: Composite fermions fill integer Landau levels.** In the reduced field $B^*$, the composite fermions themselves fill $p$ complete Landau levels (integer filling $\nu^* = p$):

$$
\nu^* = \frac{n_e h}{eB^*} = p \quad \Longrightarrow \quad \nu = \frac{n_e h}{eB} = \frac{\nu^*}{1 + 2s\nu^*} = \frac{p}{2sp + 1}.
$$

**Step 4: Jain sequence.** Setting $s=1$ and varying $p = 1, 2, 3, \ldots$ gives the principal Jain sequence

$$
\nu = \frac{1}{3},\; \frac{2}{5},\; \frac{3}{7},\; \frac{4}{9},\; \ldots
$$

observed experimentally. The $p = 1, s = 1$ case recovers the Laughlin $\nu = 1/3$ state. Negative $B^*$ (composite fermions filling $p$ levels in the opposite direction) gives the particle-hole conjugate sequence $\nu = p/(2sp - 1) = 2/3, 3/5, \ldots$.

**Step 5: Why this explains the gap.** Composite fermions in an integer number of filled Landau levels have a gap $\hbar\omega^*_c = \hbar eB^*/m^*$ to the next CF Landau level. This gap persists (with corrections from interactions) back in the original electron system—explaining the incompressible nature of FQHE states across the entire Jain sequence.
:::

### Summary

- **Fractionalization is real**: at filling $
u = 1/3$, the elementary excitations carry charge $e/3$, confirmed by experiment.
- **Statistics is emergent**: the exchange phase $	heta = \pi/3$ is not fundamental — it arises from many-body correlations in the Laughlin ground state.
- **Topology protects**: fractional charge and statistics are topological invariants, robust against local perturbations.
- The **charge-flux composite** picture (§2.3.1) reduces fractional statistics to the Aharonov-Bohm effect: each $
u = 1/m$ anyon carries charge $e/m$ and one flux quantum, giving $	heta_{\mathrm{br}} = 2\pi/m$ and $	heta_{\mathrm{ex}} = \pi/m$.
