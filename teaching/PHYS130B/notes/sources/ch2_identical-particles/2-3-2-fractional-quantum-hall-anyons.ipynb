## Lecture Notes

### The Fractional Quantum Hall Effect

When a 2D electron gas is placed in a strong perpendicular magnetic field at low temperature, the Hall conductance is quantized. At certain fractional filling factors, the system forms an incompressible quantum liquid:

:::{admonition} Key Filling Fractions
:class: important

The fractional quantum Hall effect (FQHE) occurs at filling fractions $\nu = 1/3, 2/5, 1/5, 2/3, \ldots$ The most prominent is $\nu = 1/3$, discovered by Tsui, Störmer, and Gossard in 1982 (Nobel Prize 1998).
:::

The FQHE cannot be explained by single-particle physics — it arises from strong electron-electron correlations. The ground state is described by the **Laughlin wavefunction**:

$$\psi(z_1, \ldots, z_N) = \prod_{i < j} (z_i - z_j)^m \, \mathrm{e}^{-\sum_k |z_k|^2 / 4\ell_B^2}$$

where $z_k = x_k + \mathrm{i} y_k$ are complex coordinates, $m$ is an odd integer, $\ell_B = \sqrt{\hbar/eB}$ is the magnetic length, and $\nu = 1/m$.

:::{admonition} Quasiparticle Properties at $\nu = 1/3$
:class: important

The elementary excitations (quasiparticles) of the $\nu = 1/3$ state are **anyons** with:

- **Fractional charge**: $q = e/3$ (one-third of the electron charge)
- **Fractional statistics**: statistical angle $\theta = \pi/3$ (exchange phase $\mathrm{e}^{\mathrm{i}\pi/3}$)

These properties have been confirmed experimentally through shot noise measurements (charge) and interferometry (statistics).
:::

### The Charge-Flux Composite Picture

The origin of fractional statistics can be understood through a simple and powerful picture: **each anyon is a charge bound to a magnetic flux tube**.

:::{admonition} Charge-Flux Composite
:class: important

Attach a fictitious flux tube $\Phi$ to each particle of charge $q$. When one composite moves around another, the charge picks up an **Aharonov-Bohm phase**:

$$\theta_{\text{AB}} = \frac{q \Phi}{\hbar}$$

For the exchange (half-loop), the statistical angle is:

$$\theta = \frac{1}{2} \cdot \frac{q \Phi}{\hbar}$$
:::

:::{admonition} Example: $\nu = 1/3$ Anyons
:class: note

At $\nu = 1/3$, each electron binds 3 magnetic flux quanta $\Phi_0 = h/e$. A quasihole carries charge $q = e/3$ and flux $\Phi = \Phi_0/3$. The braiding phase for taking one quasihole around another is:

$$\theta_{\text{braid}} = \frac{(e/3)(\Phi_0/3)}{\hbar} = \frac{(e/3)(h/3e)}{\hbar} = \frac{2\pi}{3}$$

Since exchange is half a braid: $\theta_{\text{exchange}} = \pi/3$. This matches the known statistical angle.
:::

This picture is powerful because it reduces the exotic phenomenon of fractional statistics to the familiar Aharonov-Bohm effect: a charge moving around a flux picks up a phase proportional to the enclosed flux.

:::{admonition} Summary
:class: note

The FQHE demonstrates that:

1. **Fractionalization is real**: electrons in 2D can collectively produce excitations with fractional charge
2. **Statistics is emergent**: the anyonic statistics of quasiparticles is not put in by hand — it emerges from the many-body ground state
3. **Topology protects**: the fractional charge and statistics are topological invariants, robust against local perturbations
:::

:::{admonition} Discussion: Is $e/3$ a Real Charge?
:class: dropdown tip

A quasihole in the $\nu = 1/3$ state carries charge $e/3$, but it is not literally one-third of an electron — it is a collective excitation of many electrons. In what sense is the fractional charge "real"? Can you measure $e/3$ in a single experiment, or only statistically?
