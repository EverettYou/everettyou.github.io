## Lecture Notes

Observation in quantum mechanics is not passive—it is fundamentally tied to the mathematical structure of reality through operators. To measure a physical quantity is to ask which values are possible, which states correspond to those values, and how certain we can be of the result. Hermitian operators encode this structure completely. They are the mathematical embodiment of our ability to know, predict, and control quantum systems, and understanding their properties is essential to understanding the bridge between theory and experiment.

### Observables as Hermitian Operators

In quantum mechanics, a physical **observable** (measurable quantity) is represented by a **Hermitian operator** acting on the Hilbert space. This is one of the fundamental postulates of quantum mechanics.

**Definition (Hermitian operator)**: An operator $\hat{O}$ is **Hermitian** (or **self-adjoint**) if

$$

\hat{O}^\dagger = \hat{O}

$$ (eq-hermitian-def)

where $\hat{O}^\dagger$ is the **Hermitian conjugate** (or **adjoint**), defined by the property

$$

\langle \psi | \hat{O} \phi \rangle = \langle \hat{O}^\dagger \psi | \phi \rangle

$$ (eq-hermitian-conjugate-def)

for all states $|\psi\rangle, |\phi\rangle$.

### Hermitian Conjugate and Matrix Form

For a matrix $M$, the Hermitian conjugate is the **conjugate transpose**:

$$

M^\dagger = (M^*)^T = (M^T)^*

$$ (eq-hermitian-conjugate-matrix)

$$

M = \begin{pmatrix} a & b \\ c & d \end{pmatrix},

$$

A matrix is Hermitian if $M = M^\dagger$. To compute the Hermitian conjugate:

:::{admonition} Derivation: Hermitian Conjugate
:class: dropdown note

$$

M^\dagger = \begin{pmatrix} a^* & c^* \\ b^* & d^* \end{pmatrix}.

$$

For Hermiticity, we need $M = M^\dagger$, which requires:
- Diagonal elements are **real**: $a, d \in \mathbb{R}$
- Off-diagonal elements are **conjugate pairs**: $b = c^*$
:::

Hermitian operators have the crucial property that their **eigenvalues are real**. Since measurement outcomes are real numbers (we observe real values on our instruments), the operators representing observables must be Hermitian. This is not just convenience—it's a requirement for quantum mechanics to be physically sensible.
:::

### Eigenvalues and Eigenvectors

For a linear operator $\hat{O}$ acting on a state $|\psi\rangle$, there may exist special states (eigenstates) that are only **rescaled** by the operator:

**Eigenvalue equation**:

$$

\hat{O} |\psi\rangle = \lambda |\psi\rangle

$$ (eq-eigenvalue-equation)

where:
- $|\psi\rangle$ is an **eigenvector** (or **eigenstate**)
- $\lambda$ is the **eigenvalue**, a number (generally complex) that scales the eigenvector

#### Properties of Hermitian Operators

**Theorem**: Let $\hat{O}$ be a Hermitian operator. Then:

1. **Eigenvalues are real**: If $\hat{O}|\psi\rangle = \lambda|\psi\rangle$ with $\hat{O}^\dagger = \hat{O}$, then $\lambda \in \mathbb{R}$.

   :::{admonition} Proof: Eigenvalues are Real
:class: dropdown note

Take the inner product of the eigenvalue equation with $\langle\psi|$:

$$\langle\psi|\hat{O}|\psi\rangle = \lambda \langle\psi|\psi\rangle = \lambda.$$

Taking the complex conjugate:
$$\langle\psi|\hat{O}|\psi\rangle^* = \lambda^*.$$

But also:

$$\langle\psi|\hat{O}|\psi\rangle^* = \langle\psi|\hat{O}^\dagger|\psi\rangle = \langle\psi|\hat{O}|\psi\rangle = \lambda$$

(using Hermiticity and the fact that $\langle\psi|\hat{O}|\psi\rangle$ is a number). Thus $\lambda = \lambda^*$, so $\lambda \in \mathbb{R}$. ✓
:::

2. **Eigenvectors are orthogonal** (for distinct eigenvalues): If $\hat{O}|\psi_1\rangle = \lambda_1|\psi_1\rangle$ and $\hat{O}|\psi_2\rangle = \lambda_2|\psi_2\rangle$ with $\lambda_1 \neq \lambda_2$, then $\langle\psi_1|\psi_2\rangle = 0$.

   :::{admonition} Proof: Eigenvectors are Orthogonal
:class: dropdown note

$$\langle\psi_1|\hat{O}|\psi_2\rangle = \lambda_2 \langle\psi_1|\psi_2\rangle.$$

Also:

$$\langle\psi_1|\hat{O}|\psi_2\rangle = (\hat{O}|\psi_1\rangle)^\dagger|\psi_2\rangle = (\lambda_1|\psi_1\rangle)^\dagger|\psi_2\rangle = \lambda_1^* \langle\psi_1|\psi_2\rangle = \lambda_1 \langle\psi_1|\psi_2\rangle$$

(since $\lambda_1$ is real for Hermitian operators). Thus:

$$\lambda_2 \langle\psi_1|\psi_2\rangle = \lambda_1 \langle\psi_1|\psi_2\rangle \implies (\lambda_2 - \lambda_1)\langle\psi_1|\psi_2\rangle = 0.$$

Since $\lambda_1 \neq \lambda_2$, we have $\langle\psi_1|\psi_2\rangle = 0$. ✓
:::

3. **Completeness**: For a 2-dimensional system (like a qubit), there are at most **two linearly independent eigenvectors**. These span the entire Hilbert space and form a **basis** (the **eigenbasis** of the operator).

### The Pauli Matrices

The most important Hermitian operators for a qubit are the **Pauli matrices**, representing spin along three orthogonal directions:

$$

\hat{\sigma}^x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}, \quad
\hat{\sigma}^y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}, \quad
\hat{\sigma}^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}

$$ (eq-pauli-matrices)

All three are Hermitian ($\hat{\sigma}^i = (\hat{\sigma}^i)^\dagger$) with eigenvalues $\pm 1$.

The Pauli matrices satisfy the algebra:

$$

[\hat{\sigma}^i, \hat{\sigma}^j] = 2i\epsilon_{ijk}\hat{\sigma}^k

$$

where $\epsilon_{ijk}$ is the Levi-Civita symbol. A detailed exploration of Pauli algebra, spin representations, and commutation relations is the focus of **§2.2 in Chapter 2**.

### Expectation Values

For a quantum state $|\psi\rangle$ and an observable $\hat{O}$, the **expectation value** (or **mean value**) is

$$

\langle \hat{O} \rangle = \langle \psi | \hat{O} | \psi \rangle

$$ (eq-expectation-value)

**Properties**:
- For a Hermitian operator, $\langle \hat{O} \rangle$ is **real** (automatically).
- It represents the **average outcome** if we measure $\hat{O}$ on many identically prepared states.

For the state $|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ and observable $\hat{\sigma}^z$:

:::{admonition} Calculation: Expectation Value
:class: dropdown note

$$

\langle \hat{\sigma}^z \rangle = \frac{1}{2}(\langle 0| + \langle 1|) \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)

$$

$$

= \frac{1}{2}(\langle 0| + \langle 1|)\frac{1}{\sqrt{2}}(|0\rangle - |1\rangle) = \frac{1}{2\sqrt{2}}(1 - 1) = 0

$$

:::

This makes physical sense: the state is an equal superposition of spin-up and spin-down along z, so the average z-spin is zero.

### The Measurement Postulate

**Postulate (Measurement)**: When we measure an observable $\hat{O}$ on a system in state $|\psi\rangle$:

1. **Possible outcomes**: The measurement result is an eigenvalue $\lambda_i$ of $\hat{O}$.
2. **Probability**: The probability of obtaining outcome $\lambda_i$ is
   $$

   P_i = |\langle \psi_i | \psi \rangle|^2

   $$ (eq-measurement-probability)

   where $|\psi_i\rangle$ is the normalized eigenvector corresponding to eigenvalue $\lambda_i$.
3. **State collapse**: After the measurement, the state **collapses** to the eigenstate $|\psi_i\rangle$ corresponding to the observed outcome.

#### Average Outcome

The **expectation value** is the average of all possible outcomes weighted by their probabilities:

$$

\langle \hat{O} \rangle = \sum_i \lambda_i P_i = \sum_i \lambda_i |\langle \psi_i | \psi \rangle|^2

$$ (eq-expectation-as-sum)

This is consistent with the definition $\langle \hat{O} \rangle = \langle \psi | \hat{O} | \psi \rangle$.

### Spectral Decomposition

Any Hermitian operator can be decomposed in terms of its eigenvalues and eigenprojectors:

$$

\hat{O} = \sum_i \lambda_i |\psi_i\rangle\langle\psi_i|

$$ (eq-spectral-decomposition)

where $|\psi_i\rangle\langle\psi_i|$ is the **projector** onto the eigenstate $|\psi_i\rangle$.

For $\hat{\sigma}^z$:

:::{admonition} Verification: Spectral Decomposition
:class: dropdown note

$$

\hat{\sigma}^z = (+1)|0\rangle\langle 0| + (-1)|1\rangle\langle 1| = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix} - \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \quad \checkmark

$$

:::

:::{admonition} Key Result: Measurement and Expectation
:class: important

The expectation value $\langle \hat{O} \rangle = \langle \psi | \hat{O} | \psi \rangle$ is:
- The **average result** of many measurements on identically prepared states.
- Computable as a **sum of eigenvalues weighted by probabilities**.
- A **real number** (when $\hat{O}$ is Hermitian).
- The **one number that characterizes the state's response to measurement** of that observable.
:::

### Summary

- **Observables** are represented by **Hermitian operators**, which have real eigenvalues and orthogonal eigenvectors.
- The **Pauli matrices** $\hat{\sigma}^x, \hat{\sigma}^y, \hat{\sigma}^z$ are the fundamental observables for a qubit, representing spin along the three axes.
- **Eigenvalues** are the possible measurement outcomes; **eigenvectors** are the states with definite outcomes.
- The **measurement postulate** connects theory to experiment: measurement outcomes are eigenvalues, with probabilities given by the Born rule.
- **Expectation values** are averages of measurement outcomes and are computed as $\langle \hat{O} \rangle = \langle \psi | \hat{O} | \psi \rangle$.
