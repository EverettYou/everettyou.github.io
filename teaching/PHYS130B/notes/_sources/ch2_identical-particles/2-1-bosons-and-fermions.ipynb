# 2.1 Bosons and Fermions

### Overview

The distinction between bosons and fermions is **fundamental** to all of physics:
- **Chemistry and atomic structure** depend on the Pauli exclusion principle.
- **Superconductivity** emerges from pairing of fermions.
- **Lasers and quantum optics** exploit boson properties (unlimited occupation in one state).
- **Dense matter** (white dwarfs, neutron stars) is supported by fermion degeneracy pressure.
- **Particle physics** connects spin and statistics (spin-statistics theorem).


| Property | Bosons | Fermions |
|----------|--------|----------|
| Spin | Integer | Half-integer |
| Exchange behavior | Symmetric | Antisymmetric |
| Occupancy | Unlimited per state | At most 1 per state |
| Algebra | Commutation relations | Anticommutation relations |
| Examples | Photons, $^4$He, phonons | Electrons, quarks, $^3$He |
| Consequence | Bose condensation possible | Degeneracy pressure |
