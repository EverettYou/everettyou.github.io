## Lecture Notes

When multiple angular momenta combine—orbital and spin, or spin and spin—their coupling rules determine the allowed energy levels and symmetries of atoms and molecules. This section reveals how the composition of angular momenta generates the rich spectroscopy of the atom and the basis for chemical bonding.
### Total Angular Momentum

When two particles (or subsystems) each carry angular momentum, the **total angular momentum** is their vector sum:

$$

\boldsymbol{J} = \boldsymbol{J}_1 + \boldsymbol{J}_2

$$ (eq-total-angmom)

The components add:

$$

J_x = J_{1x} + J_{2x}, \quad J_y = J_{1y} + J_{2y}, \quad J_z = J_{1z} + J_{2z}

$$

Both $\boldsymbol{J}_1$ and $\boldsymbol{J}_2$ satisfy angular momentum commutation relations, so their sum automatically satisfies:

$$

[J_i, J_j] = i\hbar \epsilon_{ijk} J_k

$$

### The Tensor Product Hilbert Space

If particle 1 has angular momentum quantum numbers $j_1, m_1$ and particle 2 has $j_2, m_2$, the combined system lives in a Hilbert space that is a **tensor product**:

$$

\mathcal{H}_{\text{total}} = \mathcal{H}_{j_1} \otimes \mathcal{H}_{j_2}

$$

This space has dimension $(2j_1 + 1)(2j_2 + 1)$.

:::{admonition} Triangle rule
:class: important
When adding angular momenta $j_1$ and $j_2$, the total $J$ ranges from $|j_1 - j_2|$ to $j_1 + j_2$ in integer steps. The Clebsch-Gordan coefficients give the expansion of $|J, M\rangle$ in terms of $|j_1, m_1\rangle|j_2, m_2\rangle$.
:::

### Summary

- Key takeaway 1
- Key takeaway 2
- Key takeaway 3
:::{important} Clebsch-Gordan Coefficients
:class: important

When coupling J₁ ⊗ J₂, the tensor product basis {|j₁,m₁⟩|j₂,m₂⟩} is not simultaneous eigenbasis of Ĵ_total² and Ĵ_z,total. Clebsch-Gordan coefficients convert between uncoupled and coupled bases.
:::

:::{note} Selection Rules for j
:class: note

When coupling j₁ and j₂, the total j ranges from |j₁-j₂| to j₁+j₂. For instance, 1/2 ⊗ 1/2 gives j=0 (singlet) and j=1 (triplet). For each j, m still ranges from -j to +j.
:::

:::{warning} Identical Particles and Symmetry
:class: warning

For identical particles, angular momentum states must respect exchange symmetry. Fermionic particles require opposite total symmetry: if spatial part is symmetric, spin part must be antisymmetric (singlet), and vice versa.
:::


## Uncoupled (Product) Basis

### Definition

The **uncoupled basis** (also called the **product basis**) labels states by the individual angular momenta of each particle:

$$

|j_1, m_1\rangle \otimes |j_2, m_2\rangle \equiv |j_1, m_1; j_2, m_2\rangle

$$ (eq-uncoupled-basis)

This is an **eigenbasis** of $J_{1z}$ and $J_{2z}$:

$$

J_{1z} |j_1, m_1; j_2, m_2\rangle = \hbar m_1 |j_1, m_1; j_2, m_2\rangle

$$

$$

J_{2z} |j_1, m_1; j_2, m_2\rangle = \hbar m_2 |j_1, m_1; j_2, m_2\rangle

$$

**Total $z$-component**:

$$

J_z = J_{1z} + J_{2z} \quad \Rightarrow \quad J_z |j_1, m_1; j_2, m_2\rangle = \hbar(m_1 + m_2) |j_1, m_1; j_2, m_2\rangle

$$

### Advantage and Limitation

**Advantage**: Easy to construct and compute; natural for certain problems (e.g., when spin-orbit coupling is weak).

**Limitation**: $|j_1, m_1; j_2, m_2\rangle$ is **not** an eigenstate of $J^2$ (the total angular momentum squared). So if we want to classify states by their total angular momentum quantum number $j$, we must work in a different basis.

## Coupled Basis

### Definition

The **coupled basis** (or **eigenbasis of total angular momentum**) labels states by the total quantum numbers $j$ and $m$:

$$

|j_1, j_2; j, m\rangle

$$ (eq-coupled-basis)

where $j$ and $m$ are eigenvalues of $J^2$ and $J_z$:

$$

J^2 |j_1, j_2; j, m\rangle = \hbar^2 j(j+1) |j_1, j_2; j, m\rangle

$$

$$

J_z |j_1, j_2; j, m\rangle = \hbar m |j_1, j_2; j, m\rangle

$$

### Triangle Rule: Allowed Values of $j$

Not all combinations of $j_1, j_2$ with $j$ are possible. The constraint is the **triangle rule**:

$$

j \in \{|j_1 - j_2|, |j_1 - j_2| + 1, \ldots, j_1 + j_2 - 1, j_1 + j_2\}

$$ (eq-triangle-rule)

**Physical interpretation**: The total angular momentum must satisfy a triangle inequality—the same rule that governs vectors in classical mechanics.

For each allowed $j$, the magnetic quantum number ranges as:

$$

m \in \{-j, -j+1, \ldots, j-1, j\}

$$

:::{admonition} Derivation: Triangle Rule for Angular Momentum Addition
:class: dropdown note

The triangle rule $j \in \{|j_1 - j_2|, |j_1 - j_2| + 1, \ldots, j_1 + j_2\}$ follows from the requirement that quantum numbers must be integers or half-integers.

**Key constraint:** For a given value of $j$, the magnetic quantum number must satisfy:

$$

m \in \{-j, -j+1, \ldots, j-1, j\}

$$

In the uncoupled basis, the total $z$-component is:

$$

m_{\text{total}} = m_1 + m_2

$$

**Maximum value of $m_{\text{total}}$:**

$$

m_{\text{max}} = j_1 + j_2

$$

For a coupled state $|j_1, j_2; j, m\rangle$, the maximum value of $m$ is $j$ itself. Therefore:

$$

j_{\text{max}} = j_1 + j_2

$$

**Minimum value of $m_{\text{total}}$:**

$$

m_{\text{min}} = -j_1 - j_2

$$

This also constrains $j_{\text{min}}$. However, there is another constraint: the total dimension must match.

**Dimension matching:**

- Uncoupled basis: $(2j_1 + 1)(2j_2 + 1)$ states
- Coupled basis: $\sum_{j} (2j + 1)$ where $j$ ranges over allowed values

These must be equal. With the triangle rule, we get exactly the right count:

$$

\sum_{j=|j_1-j_2|}^{j_1+j_2} (2j + 1) = (2j_1 + 1)(2j_2 + 1) \quad ✓

$$

Thus the minimum is $j_{\text{min}} = |j_1 - j_2|$, giving the full range:

$$

j \in \{|j_1 - j_2|, |j_1 - j_2| + 1, \ldots, j_1 + j_2\}

$$

:::



:::{admonition} Example: Two Spin-1/2 Particles
:class: example


For $j_1 = 1/2$ and $j_2 = 1/2$:

$$

j \in \{|1/2 - 1/2|, \ldots, 1/2 + 1/2\} = \{0, 1\}

$$

So we can have:
- **Total singlet** ($j = 0$): 1 state
- **Total triplet** ($j = 1$): 3 states

Total: $1 + 3 = 4$ states, which matches $(2 \cdot 1/2 + 1)(2 \cdot 1/2 + 1) = 2 \times 2 = 4$. ✓
:::

## Clebsch-Gordan Coefficients

### The Transformation

The coupled and uncoupled bases span the same space but differ in which observable they diagonalize. To transform between them, we use a **unitary transformation** with coefficients called **Clebsch-Gordan coefficients** (CG coefficients):

$$

|j_1, j_2; j, m\rangle = \sum_{m_1, m_2} C^{j, m}_{j_1, m_1; j_2, m_2} |j_1, m_1; j_2, m_2\rangle

$$ (eq-clebsch-gordan-forward)

The inverse transformation:

$$

|j_1, m_1; j_2, m_2\rangle = \sum_{j} C^{j, m_1+m_2}_{j_1, m_1; j_2, m_2} |j_1, j_2; j, m_1 + m_2\rangle

$$ (eq-clebsch-gordan-inverse)

where the summation is over all allowed values of $j$ (by the triangle rule).

### Properties

1. **Orthonormality**:
   $$

   \sum_{m_1, m_2} C^{j, m}_{j_1, m_1; j_2, m_2} C^{j', m'}_{j_1, m_1; j_2, m_2} = \delta_{jj'} \delta_{mm'}

   $$

2. **Selection rules** (zeros):
   - $C^{j, m}_{j_1, m_1; j_2, m_2} = 0$ unless $m = m_1 + m_2$ (conservation of $m$).
   - $C^{j, m}_{j_1, m_1; j_2, m_2} = 0$ unless $|j_1 - j_2| \leq j \leq j_1 + j_2$ (triangle rule).

3. **Realness**: The CG coefficients are **real numbers** (not complex).

4. **Symmetries**: CG coefficients satisfy various symmetry relations that reduce computation time.

### Computational Access

CG coefficients are **not** simple algebraic expressions (except for small $j$ values). They are typically:
- **Tabulated** in references (e.g., Landau-Lifshitz, Varshalovich et al.).
- **Computed** by computer algebra systems (Mathematica, SymPy, Wolfram Language).
- **Evaluated** online via databases (NIST, etc.).

For simple cases, they can be derived algebraically (see examples below).:::{admonition} Derivation: Orthonormality of Clebsch-Gordan Coefficients
:class: dropdown note

The Clebsch-Gordan coefficients form a **unitary transformation** between orthonormal bases, so they must satisfy orthonormality relations.

**Derivation:**

Both $|j_1, m_1; j_2, m_2\rangle$ and $|j_1, j_2; j, m\rangle$ are orthonormal bases:

$$

\langle j_1, m_1'; j_2, m_2' | j_1, m_1; j_2, m_2 \rangle = \delta_{m_1' m_1} \delta_{m_2' m_2}

$$

$$

\langle j_1, j_2; j', m' | j_1, j_2; j, m \rangle = \delta_{j'j} \delta_{m'm}

$$

Expanding using CG coefficients:

$$

|j_1, j_2; j, m\rangle = \sum_{m_1, m_2} C^{j,m}_{j_1, m_1; j_2, m_2} |j_1, m_1; j_2, m_2\rangle

$$

Take the inner product of two such states:

$$

\langle j_1, j_2; j', m' | j_1, j_2; j, m \rangle = \sum_{m_1, m_2} \sum_{m_1', m_2'} (C^{j',m'}_{j_1, m_1'; j_2, m_2'})^* C^{j,m}_{j_1, m_1; j_2, m_2} \langle j_1, m_1'; j_2, m_2' | j_1, m_1; j_2, m_2 \rangle

$$

$$

= \sum_{m_1, m_2} (C^{j',m'}_{j_1, m_1; j_2, m_2})^* C^{j,m}_{j_1, m_1; j_2, m_2}

$$

Since CG coefficients are **real**:

$$

\delta_{j'j} \delta_{m'm} = \sum_{m_1, m_2} C^{j',m'}_{j_1, m_1; j_2, m_2} C^{j,m}_{j_1, m_1; j_2, m_2}

$$

This is the **orthonormality relation** for CG coefficients. It encodes:
- Selection rule: $C = 0$ unless $m = m_1 + m_2$
- Normalization: $\sum_{m_1} [C^{j,m}_{j_1, m_1; j_2, m_2}]^2 = 1$ (for fixed $j, m, j_2, m_2$)
:::


## Two Spin-1/2 Particles: Explicit Construction

### The Four Uncoupled Basis States

For two spin-1/2 particles, the uncoupled basis has four states:

$$

|\uparrow\uparrow\rangle = |+\rangle_1 |+\rangle_2, \quad
|\uparrow\downarrow\rangle = |+\rangle_1 |-\rangle_2, \quad
|\downarrow\uparrow\rangle = |-\rangle_1 |+\rangle_2, \quad
|\downarrow\downarrow\rangle = |-\rangle_1 |-\rangle_2

$$

where $|+\rangle_i = |1/2, +1/2\rangle_i$ denotes spin-up on particle $i$, and similarly for $|-\rangle_i$.

### The Triplet States ($j = 1, m = 1, 0, -1$)

The **triplet** ($j = 1$) is the **symmetric** combination:

$$

|1, 1\rangle = |\uparrow\uparrow\rangle

$$

$$

|1, 0\rangle = \frac{1}{\sqrt{2}} (|\uparrow\downarrow\rangle + |\downarrow\uparrow\rangle)

$$

$$

|1, -1\rangle = |\downarrow\downarrow\rangle

$$ (eq-triplet-states)

Each triplet state is **symmetric** under particle exchange: $P_{12} |1, m\rangle = |1, m\rangle$.

### The Singlet State ($j = 0, m = 0$)

The **singlet** ($j = 0$) is the **antisymmetric** combination:

$$

|0, 0\rangle = \frac{1}{\sqrt{2}} (|\uparrow\downarrow\rangle - |\downarrow\uparrow\rangle)

$$ (eq-singlet-state)

The singlet state is **antisymmetric** under particle exchange: $P_{12} |0, 0\rangle = -|0, 0\rangle$.

### Verification

Let's verify that $|1, 0\rangle$ is an eigenstate of $J^2$:

$$

J^2 = (\boldsymbol{J}_1 + \boldsymbol{J}_2)^2 = J_1^2 + J_2^2 + 2\boldsymbol{J}_1 \cdot \boldsymbol{J}_2

$$

::{admonition} Derivation: Eigenstate Verification
:class: dropdown note

For each spin-1/2: $J_i^2 |1/2, m_i\rangle = \frac{3\hbar^2}{4} |1/2, m_i\rangle$.

$$

\boldsymbol{J}_1 \cdot \boldsymbol{J}_2 = J_{1z} J_{2z} + \frac{1}{2}(J_{1+} J_{2-} + J_{1-} J_{2+})

$$

Applied to $|1, 0\rangle = \frac{1}{\sqrt{2}}(|\uparrow\downarrow\rangle + |\downarrow\uparrow\rangle)$:

$$

J^2 |1, 0\rangle = 2 \hbar^2 + \text{(action of } \boldsymbol{J}_1 \cdot \boldsymbol{J}_2\text{)} = 2\hbar^2 |1, 0\rangle

$$

This confirms $J^2 = \hbar^2 j(j+1) = 2\hbar^2$ for $j = 1$. ✓
:::

## Physical Implications: Singlet and Triplet States

### Spin Correlations in the Singlet

The singlet state $|0, 0\rangle = \frac{1}{\sqrt{2}}(|\uparrow\downarrow\rangle - |\downarrow\uparrow\rangle)$ has a remarkable property:

**If we measure the spin of particle 1 along $z$ and find $\uparrow$, then particle 2 is **guaranteed** to be $\downarrow$** (and vice versa).

This is **maximal entanglement**: the two spins are perfectly anticorrelated.

### Experimental Test: Bell's Inequality

The singlet state can be used to test **Bell's inequality**, demonstrating that quantum mechanics exhibits stronger correlations than any classical hidden-variable theory allows. This is a fundamental test of the non-classical nature of quantum mechanics.

### In Atomic and Molecular Physics

**Helium atom ground state**: Two electrons in the $1s$ orbital must form a spin singlet (to satisfy the Pauli exclusion principle, as the spatial part is symmetric).

$$

\Psi_{He, \text{ground}} = \phi_{1s}(\boldsymbol{r}_1) \phi_{1s}(\boldsymbol{r}_2) \times |0, 0\rangle

$$

The two electron spins are paired (antiparallel).

**Hydrogen molecule** $H_2$: Two electrons can pair in a singlet state, lowering the energy and stabilizing the bond.

## Spin-Orbit Coupling

### The Interaction

An electron moving in an electromagnetic field experiences a **spin-orbit interaction** proportional to:

$$

H_{SO} = \lambda \boldsymbol{L} \cdot \boldsymbol{S}

$$ (eq-spinorbit-coupling)

where $\boldsymbol{L}$ is orbital angular momentum, $\boldsymbol{S}$ is spin, and $\lambda$ is a coupling constant (depends on the electric field gradient).

### Importance of the Coupled Basis

The spin-orbit interaction $\boldsymbol{L} \cdot \boldsymbol{S}$ is **diagonal** in the coupled basis $|\ell, s; j, m_j\rangle$ but **not** in the uncoupled basis $|\ell, m_\ell; s, m_s\rangle$.

Using coupled basis states simplifies the problem enormously: the uncoupled states mix significantly due to $\boldsymbol{L} \cdot \boldsymbol{S}$.

### Fine Structure

For an electron in an atom with orbital angular momentum $\ell$:

$$

j \in \{\ell - 1/2, \ell + 1/2\} \quad \text{(for } \ell \geq 1\text{)}

$$

The energy splitting between $j = \ell + 1/2$ and $j = \ell - 1/2$ is the **fine structure**—a small but measurable correction to atomic energies.

**Example**: Hydrogen $2p$ level splits into $2p_{3/2}$ and $2p_{1/2}$.

## Wigner-Eckart Theorem (Brief Mention)

### General Principle

When working with coupled angular momentum states, matrix elements of tensor operators can be related to a **reduced matrix element** and a **geometric factor** (a CG coefficient):

$$

\langle j_1', m_1'; j_2', m_2' | T^k_q | j_1, m_1; j_2, m_2 \rangle \propto C^{j'}_{j, k; m_1 + m_2 - q} \langle j_1' j_2' j' || T^k || j_1 j_2 j \rangle

$$

where $T^k_q$ is a tensor operator of rank $k$.

### Power

This theorem is powerful because it allows us to:
- Reduce complex calculations to a few reduced matrix elements.
- Apply selection rules systematically.
- Compute transition rates in atomic spectroscopy.

It is widely used in atomic and nuclear physics.

## Summary

- **Total angular momentum**: $\boldsymbol{J} = \boldsymbol{J}_1 + \boldsymbol{J}_2$; satisfies same commutation relations.
- **Uncoupled basis**: $|j_1, m_1; j_2, m_2\rangle$ diagonalizes $J_{1z}, J_{2z}$ but not $J^2$.
- **Coupled basis**: $|j_1, j_2; j, m\rangle$ diagonalizes $J^2, J_z$.
- **Triangle rule**: $j \in \{|j_1 - j_2|, \ldots, j_1 + j_2\}$.
- **Clebsch-Gordan coefficients**: Real unitary transformation between bases; tabulated in references.
- **Two spin-1/2 particles**: Singlet ($j=0$, antisymmetric) and triplet ($j=1$, symmetric) states.
- **Spin-orbit coupling** $\boldsymbol{L} \cdot \boldsymbol{S}$ is naturally diagonal in coupled basis; causes fine structure in atoms.
- **Wigner-Eckart theorem**: Relates matrix elements to reduced elements and geometric factors.
