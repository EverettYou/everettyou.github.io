# 5.2 Time-Dependent Perturbation Theory

### Overview

When a quantum system is driven by a time-varying external field, transitions between energy levels occur. Time-dependent perturbation theory provides the framework for computing transition rates, leading to Fermi's golden rule—the workhorse formula for decay rates, absorption spectra, and scattering cross sections across all of physics.
