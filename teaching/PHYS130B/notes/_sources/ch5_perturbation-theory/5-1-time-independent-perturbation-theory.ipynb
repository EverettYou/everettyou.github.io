# 5.1 Time-Independent Perturbation Theory

### Overview

**Perturbation theory** is a systematic method for extending knowledge of exactly solvable quantum systems to nearby regimes in parameter space. Given the spectrum of $H_0$ and a small perturbation $V$, we find the spectrum of $H(\lambda) = H_0 + \lambda V$ as a power series in $\lambda$.
