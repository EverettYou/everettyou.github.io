## Lecture Notes

A quantum channel is a map that transforms one density matrix into another while respecting the rules of quantum mechanics: probabilities must be conserved, and all quantum properties preserved even for subsystems. These linear maps are the quantum generalization of classical communication channels, encoding everything from the evolution of open systems to the effects of quantum noise.

:class: important

 Quantum Channel

A **quantum channel** (or **quantum operation**) is a linear map:

$$


\mathcal{E}: \text{Density matrices on } \mathcal{H}_\text{in} \to \text{Density matrices on } \mathcal{H}_\text{out}

$$


Representing any physical process: unitary evolution, measurement, dissipation, entanglement with environment, etc.

### CPTP Properties

A valid quantum channel must satisfy:

**1. Linearity**:

$$


\mathcal{E}(p\rho_1 + q\rho_2) = p\mathcal{E}(\rho_1) + q\mathcal{E}(\rho_2)

$$ (eq-channel-linearity)

**2. Trace preservation**:

$$


\text{Tr}(\mathcal{E}(\rho)) = \text{Tr}(\rho) = 1

$$ (eq-channel-trace-preserving)

Probabilities are conserved; output is a valid density matrix.

**3. Positivity**:

$$


\rho \geq 0 \quad \Rightarrow \quad \mathcal{E}(\rho) \geq 0

$$ (eq-channel-positivity)

Positive semidefinite inputs map to positive semidefinite outputs.

**4. Complete positivity**:

$$


\mathcal{E} \otimes \mathcal{I}_n \text{ is positive for all } n

$$ (eq-channel-completely-positive)

Even when the system is entangled with auxiliary systems (e.g., reference qubits), the map preserves positivity.

#### Physical Motivation for Complete Positivity

Consider two qubits A and B, entangled. You apply a channel $\mathcal{E}$ to A only. If $\mathcal{E}$ is merely positive but **not** completely positive, the map $\mathcal{E} \otimes \mathcal{I}_B$ could produce **negative eigenvalues** in the joint system (negative probability!). This is unphysical.

**Therefore**: All physical channels must be completely positive.

### Kraus Representation Theorem

**Theorem**: A linear map $\mathcal{E}$ on density matrices is CPTP if and only if it can be written in **Kraus form**:

$$


\mathcal{E}(\rho) = \sum_{k=1}^r K_k \rho K_k^\dagger

$$ (eq-channel-kraus)

where the **Kraus operators** $\{K_k\}$ satisfy:

$$


\sum_k K_k^\dagger K_k = I

$$ (eq-channel-kraus-constraint)

(sum of products of Kraus operators equals identity).

The Kraus representation ensures complete positivity:

:::{admonition} Proof: Kraus Representation Works
:class: dropdown note

- **Trace-preserving**: $\text{Tr}(\sum_k K_k \rho K_k^\dagger) = \sum_k \text{Tr}(\rho K_k^\dagger K_k) = \text{Tr}(\rho \sum_k K_k^\dagger K_k) = \text{Tr}(\rho) = 1$. ✓

- **Positive semidefinite output**: If $\rho \geq 0$, then $\mathcal{E}(\rho) = \sum_k K_k \rho K_k^\dagger \geq 0$ (sum of positive operators). ✓

- **Completely positive**: Structure ensures positivity even under tensor product with identity. ✓
:::
#### Non-Uniqueness of Kraus Operators

Given one set $\{K_k\}$, other valid sets exist related by unitary rotation:

$$


K_k' = \sum_j U_{kj} K_j

$$


where $\{U_{kj}\}$ is a unitary matrix. Different representations yield the same channel map.

#### Minimal Kraus Representation

The **minimum number** of Kraus operators needed is $r = \text{rank}(J_\mathcal{E})$, where $J_\mathcal{E}$ is the Choi matrix (discussed below). This is called the **Kraus rank**.

:::{admonition} Choi-Jamiołkowski Isomorphism
:class: important

The **Choi matrix** (or **Choi state**) is a bipartite state that uniquely encodes a channel:

$$


J_\mathcal{E} = (\mathcal{E} \otimes \mathcal{I})(|\Phi^+\rangle\langle\Phi^+|)

$$ (eq-channel-choi)

where $|\Phi^+\rangle = \frac{1}{\sqrt{d}}\sum_{i=0}^{d-1}|ii\rangle$ is the **maximally entangled state** (with $d = \dim(\mathcal{H}_\text{in})$).

A channel $\mathcal{E}$ is CPTP if and only if its Choi matrix is positive semidefinite. This establishes a one-to-one correspondence between CPTP channels and positive semidefinite bipartite states.

:::

**Interpretation**: The identity channel applied to one half of a maximally entangled pair; the other half undergoes channel $\mathcal{E}$. The resulting state encodes the full information about $\mathcal{E}$.

#### Reconstruction

From the Choi matrix, reconstruct the channel via:

$$


\mathcal{E}(\rho) = \text{Tr}_1((I \otimes \rho^T) J_\mathcal{E})

$$ (eq-channel-choi-recovery)

where the partial transpose $\rho^T$ is applied to the input system, and we trace over the reference system.

Complete positivity can be checked via the Choi matrix:

:::{admonition} Criterion: CPTP via Choi-Jamiołkowski Matrix
:class: dropdown note

$\mathcal{E}$ is CPTP if and only if its Choi matrix $J_\mathcal{E}$ is **positive semidefinite** (eigenvalues $\geq 0$).

**Consequence**: One-to-one correspondence between CPTP channels and positive semidefinite bipartite states.
:::
### Common Quantum Channels

#### 1. Unitary Channel

$$


\mathcal{E}_U(\rho) = U \rho U^\dagger

$$ (eq-channel-unitary)

**Kraus form**: Single operator $K = U$ (with $U^\dagger U = I$).

**Properties**:
- Reversible: inverse is $\mathcal{E}_{U^\dagger}$.
- No information loss: Entropy conserved: $S(\mathcal{E}_U(\rho)) = S(\rho)$.
- Isometry in quantum information sense.

#### 2. Depolarizing Channel

With probability $p$, the state becomes completely random; with probability $1-p$, it's unchanged:

$$


\mathcal{E}_\text{depol}(\rho) = (1-p)\rho + p\frac{I}{d}

$$ (eq-channel-depolarizing)

where $d = \dim(\mathcal{H})$.

**Kraus operators** (for qubit, $d=2$):

$$


K_0 = \sqrt{1-p}\, I, \quad K_1 = \sqrt{p/3}\, \hat{\sigma}^x, \quad K_2 = \sqrt{p/3}\, \hat{\sigma}^y, \quad K_3 = \sqrt{p/3}\, \hat{\sigma}^z

$$ (eq-channel-depol-kraus)

**Physical interpretation**: Errors on all three Pauli axes with equal probability; models generic noise.

#### 3. Amplitude Damping ($T_1$ Process)

Excited state decays to ground state at rate $\Gamma_1 = 1/T_1$:

$$


\mathcal{E}_\text{amp}(\rho) = K_0 \rho K_0^\dagger + K_1 \rho K_1^\dagger

$$ (eq-channel-amplitude-damping)

with:

$$


K_0 = \begin{pmatrix} 1 & 0 \\ 0 & \sqrt{1-\gamma} \end{pmatrix}, \quad K_1 = \begin{pmatrix} 0 & \sqrt{\gamma} \\ 0 & 0 \end{pmatrix}

$$ (eq-channel-amp-kraus)

where $\gamma = 1 - \mathrm{e}^{-t/T_1}$ is the decay probability.

**Physical interpretation**: Energy dissipation; ground state is absorbing (fixed point).

#### 4. Phase Damping ($T_2$ Process, Dephasing)

Coherences decay without energy loss at rate $\Gamma_\phi = 1/T_2$:

$$


\mathcal{E}_\text{phase}(\rho) = K_0 \rho K_0^\dagger + K_1 \rho K_1^\dagger

$$ (eq-channel-phase-damping)

with:

$$


K_0 = \begin{pmatrix} 1 & 0 \\ 0 & \sqrt{1-\gamma} \end{pmatrix}, \quad K_1 = \begin{pmatrix} 0 & 0 \\ 0 & \sqrt{\gamma} \end{pmatrix}

$$ (eq-channel-phase-kraus)

where $\gamma = 1 - \mathrm{e}^{-t/T_2}$.

**Physical interpretation**: Random phase fluctuations; diagonal density matrix (classical-like).

#### 5. Erasure Channel

With probability $p$, the system is lost (replaced by flag state $|e\rangle$); with probability $1-p$, passed unchanged:

$$


\mathcal{E}_\text{erase}(\rho) = (1-p)\rho \oplus p|e\rangle\langle e|

$$ (eq-channel-erasure)

**Kraus operators**:

$$


K_0 = \sqrt{1-p}\, I, \quad K_1 = \sqrt{p}\, |e\rangle\langle 0| + \ldots

$$


**Physical interpretation**: Photon loss, quantum memory readout failure; allows identifying lost systems.

### Reversibility and Irreversibility

**Reversible (Unitary) Channel**:
- Single Kraus operator $K = U$ (unitary).
- Entropy conserved: $S(\mathcal{E}(\rho)) = S(\rho)$.
- Inverse channel exists: $\mathcal{E}^{-1}(\sigma) = U^\dagger \sigma U$.
- No information loss; state can be recovered perfectly.

**Irreversible (Non-Unitary) Channel**:
- Multiple Kraus operators ($r > 1$) or complex behavior.
- Entropy generally increases: $S(\mathcal{E}(\rho)) > S(\rho)$ (output more mixed).
- No inverse: cannot recover original state from output (information lost).
- Examples: Depolarizing, amplitude damping, erasure.

### Partial Transpose and Physical Channels

The **partial transpose** of a bipartite state $\rho_{AB}$ (w.r.t. system B) is:

$$


\rho_{AB}^{T_B} = (I_A \otimes T_B)(\rho_{AB})

$$


where $T_B$ is the transpose operation on subsystem B: $(|i\rangle\langle j|)^T = |j\rangle\langle i|$.

**Criterion for CPTP channels** (via Choi matrix):

The channel $\mathcal{E}$ is CPTP if and only if the partial transpose of its Choi matrix is positive semidefinite:

$$


J_\mathcal{E}^{T_2} \geq 0

$$ (eq-channel-partial-transpose-criterion)

**Physical interpretation**: The partial transpose detects whether a state is separable (no entanglement); similarly, it detects whether a channel is physical.

### Channel Composition

If $\mathcal{E}_1$ and $\mathcal{E}_2$ are channels, their **composition** (apply $\mathcal{E}_1$ first, then $\mathcal{E}_2$) is also a channel:

$$


(\mathcal{E}_2 \circ \mathcal{E}_1)(\rho) = \mathcal{E}_2(\mathcal{E}_1(\rho))

$$ (eq-channel-composition)

**Kraus operators**: If $\mathcal{E}_1(\rho) = \sum_k K_k \rho K_k^\dagger$ and $\mathcal{E}_2(\rho) = \sum_j L_j \rho L_j^\dagger$, then:

$$


(\mathcal{E}_2 \circ \mathcal{E}_1)(\rho) = \sum_{j,k} (L_j K_k) \rho (L_j K_k)^\dagger

$$


Kraus operators of the composition are products $L_j K_k$.

### Channel Capacity (Outlook)

The **classical capacity** of a quantum channel is the maximum rate of classical information that can be reliably transmitted:

$$


C(\mathcal{E}) = \max_{\{p_i, \rho_i\}} I(X : Y)

$$


where the max is over all input ensembles $\{p_i, \rho_i\}$, and $I(X:Y)$ is mutual information.

**Computing capacity** is generally hard; known for specific channels (depolarizing, erasure, amplitude damping).

:::{admonition} CPTP maps are the most general evolution
:class: important
A quantum channel $\mathcal{E}(\rho) = \sum_k K_k \rho K_k^\dagger$ (Kraus form) is the most general physical evolution of a density matrix. Unitary evolution is the special case with a single Kraus operator. Every open-system evolution is a quantum channel.
:::

### Summary

- **Quantum channel**: CPTP linear map from density matrices to density matrices.
- **CPTP**: Completely positive, trace-preserving; ensures valid quantum states and probabilities.
- **Kraus representation** $\mathcal{E}(\rho) = \sum_k K_k \rho K_k^\dagger$ with $\sum_k K_k^\dagger K_k = I$: universal form for channels; non-unique.
- **Choi-Jamiołkowski isomorphism**: Channel ↔ Choi state $J_\mathcal{E} = (\mathcal{E} \otimes I)(|\Phi^+\rangle\langle\Phi^+|)$; CPTP ⟺ $J_\mathcal{E} \geq 0$.
- **Common channels**: Unitary (reversible), depolarizing (generic noise), amplitude damping ($T_1$), phase damping ($T_2$), erasure.
- **Reversibility**: Unitary channels conserve entropy; non-unitary channels increase entropy (information loss).
- **Partial transpose criterion**: Channel CPTP ⟺ $J_\mathcal{E}^{T_2} \geq 0$ (partial transpose of Choi state positive).
- **Channel composition**: Sequential channels are channels; Kraus operators compose as products.
