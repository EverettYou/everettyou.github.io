## Lecture Notes

Quantum information is fundamentally different from classical bits: you cannot copy an unknown quantum state, yet you can send an unknown state across the world using only entanglement and classical communication. These paradoxes dissolve when you recognize that quantum information is a resource all its own—one that enables feats impossible in any classical universe.

### No-Cloning Theorem

**Theorem**: There is no universal unitary operator that can copy an arbitrary quantum state:

$$

U(|\psi\rangle \otimes |0\rangle) = |\psi\rangle \otimes |\psi\rangle

$$

for all $|\psi\rangle$ without prior knowledge of $|\psi\rangle$.

The proof is elegant:

:::{admonition} Proof: No-Cloning Theorem
:class: dropdown note

Suppose such a unitary $U$ exists. Test on two orthogonal states $|0\rangle$ and $|1\rangle$:

$$

U(|0\rangle \otimes |0\rangle_\text{blank}) = |0\rangle \otimes |0\rangle \quad \text{(clone succeeded)}

$$

$$

U(|1\rangle \otimes |0\rangle_\text{blank}) = |1\rangle \otimes |1\rangle \quad \text{(clone succeeded)}

$$

Now test on superposition $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$:

By linearity:

$$

U(|+\rangle \otimes |0\rangle) = \frac{1}{\sqrt{2}}(|0\rangle \otimes |0\rangle + |1\rangle \otimes |1\rangle)

$$

But for cloning to work on $|+\rangle$:

$$

U(|+\rangle \otimes |0\rangle) \stackrel{?}{=} |+\rangle \otimes |+\rangle = \frac{1}{2}(|0\rangle \otimes |0\rangle + |0\rangle \otimes |1\rangle + |1\rangle \otimes |0\rangle + |1\rangle \otimes |1\rangle)

$$

These are **not equal** (different entanglement structure). **Contradiction!** Therefore, no such universal $U$ exists.
:::
#### Consequences

- Cannot make backup copies of quantum states for protection.
- Quantum error correction requires **redundant encoding** (spreading data across many physical qubits), not direct cloning.
- Single unknown quantum state cannot be reliably measured or characterized without destroying it.

### Quantum Teleportation Protocol

**Goal**: Alice has unknown state $|\psi\rangle$; send it to Bob using entanglement + classical communication.

**Requirements**:
- Alice and Bob share a **Bell pair** (entangled state), e.g., $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|{\uparrow\uparrow}\rangle + |{\downarrow\downarrow}\rangle)$.
- Classical channel for Alice to send 2 bits to Bob.
- Alice **destroys** her copy via measurement (no-cloning consequence).

#### Protocol Steps

**Initial state**: Alice holds qubit $|\psi\rangle = \alpha|{\uparrow}\rangle + \beta|{\downarrow}\rangle$; shared Bell pair between Alice (A) and Bob (B):

$$

|\Psi_\text{initial}\rangle = |\psi\rangle_A |\Phi^+\rangle_{AB}

$$

**Step 1**: Alice performs **Bell measurement** (joint measurement of her qubit and her half of Bell pair) in the Bell basis, obtaining 2 classical bits $(b_1, b_2)$.

**Step 2**: Alice sends 2 bits to Bob classically.

**Step 3**: Based on $(b_1, b_2)$, Bob applies a **Pauli correction** (one of $I, \hat{\sigma}_x, \hat{\sigma}_y, \hat{\sigma}_z$) to his half:

$$

\text{If } (b_1, b_2) = (0,0) \Rightarrow \text{Apply } I

$$

$$

\text{If } (b_1, b_2) = (0,1) \Rightarrow \text{Apply } \hat{\sigma}_z

$$

$$

\text{If } (b_1, b_2) = (1,0) \Rightarrow \text{Apply } \hat{\sigma}_x

$$

$$

\text{If } (b_1, b_2) = (1,1) \Rightarrow \text{Apply } \hat{\sigma}_x \hat{\sigma}_z

$$

**Result**: Bob's qubit is now in state $|\psi\rangle$.

#### Key Features

- **No faster-than-light communication**: Requires 2 classical bits (limited by light speed).
- **Requires pre-shared entanglement**: Cannot work without Bell pair (teleportation requires pre-existing resource).
- **Destroys original**: Alice's qubit is measured; original state no longer exists (no-cloning respected).
- **Unknown state**: Works for any unknown state (protocol doesn't depend on $|\psi\rangle$).

### Quantum Error Correction: 3-Qubit Bit-Flip Code

**Goal**: Protect one logical qubit from single-qubit bit-flip errors using 3 physical qubits.

**Encoding scheme**: One logical qubit is encoded across three physical qubits by replicating the state:

$$

|0\rangle_L = |000\rangle \quad \text{(three identical qubits in $|0\rangle$)}

$$ (eq-qec-3qubit-0)

$$

|1\rangle_L = |111\rangle \quad \text{(three identical qubits in $|1\rangle$)}

$$ (eq-qec-3qubit-1)

A general logical state is:

$$

|\psi\rangle_L = \alpha |000\rangle + \beta |111\rangle

$$

**How it works**: If a single physical qubit flips, the resulting state (e.g., $|100\rangle$ or $|110\rangle$) differs from both logical codewords in a detectable pattern. By measuring **parity checks** (non-destructive measurements that reveal which qubit flipped without revealing the logical data), we can identify and correct the error.

**Key properties**:
- Protects against **single bit-flip errors** on any of the three qubits
- Uses **syndrome measurement** to diagnose which qubit failed
- Applies a **correction** based on the syndrome
- Does NOT measure the logical state directly (avoiding collapse)

:::{admonition} Construction: 3-Qubit Bit-Flip Code
:class: dropdown note

**Detailed encoding:**

Logical state $|\psi\rangle_L = \alpha |0\rangle_L + \beta |1\rangle_L$ is encoded by mapping:

$$

|0\rangle_L = |000\rangle \quad \text{(three identical qubits in $|0\rangle$)}

$$

$$

|1\rangle_L = |111\rangle \quad \text{(three identical qubits in $|1\rangle$)}

$$

$$

|\psi\rangle_L = \alpha |000\rangle + \beta |111\rangle

$$

**Syndrome measurement** (parity checks):

$$

P_1 = Z_1 Z_2 \quad \text{(parity of qubits 1 and 2)}

$$ (eq-qec-parity-check-1)

$$

P_2 = Z_2 Z_3 \quad \text{(parity of qubits 2 and 3)}

$$

These measure **which qubits have flipped** without collapsing the logical state.

**Syndrome table**:

| Error Location | $P_1$ outcome | $P_2$ outcome | Syndrome (s_1, s_2) |
|---|---|---|---|
| None | +1 | +1 | (0,0) |
| Qubit 1 | -1 | +1 | (1,0) |
| Qubit 2 | -1 | -1 | (1,1) |
| Qubit 3 | +1 | -1 | (0,1) |

**Recovery step**: Based on syndrome $(s_1, s_2)$, apply correction:
- $(0,0)$ → No error; apply $I$.
- $(1,0)$ → Error on qubit 1; apply $\hat{\sigma}_x^{(1)}$.
- $(1,1)$ → Error on qubit 2; apply $\hat{\sigma}_x^{(2)}$.
- $(0,1)$ → Error on qubit 3; apply $\hat{\sigma}_x^{(3)}$.

:::
#### Error Scenarios

If **one qubit** suffers a bit-flip error (e.g., $\hat{\sigma}_x$ on qubit 1):

- Error on qubit 1: $|000\rangle \to |100\rangle$, $|111\rangle \to |011\rangle$.
- Error on qubit 2: $|000\rangle \to |010\rangle$, $|111\rangle \to |101\rangle$.
- Error on qubit 3: $|000\rangle \to |001\rangle$, $|111\rangle \to |110\rangle$.

#### Syndrome Measurement

Measure **parity checks** (without measuring logical data):

$$

P_1 = Z_1 Z_2 \quad \text{(parity of qubits 1 and 2)}

$$ (eq-qec-parity-check-1)

$$

P_2 = Z_2 Z_3 \quad \text{(parity of qubits 2 and 3)}

$$

These measure **which qubits have flipped** without collapsing the logical state:

| Error Location | $P_1$ outcome | $P_2$ outcome | Syndrome (s_1, s_2) |
|---|---|---|---|
| None | +1 | +1 | (0,0) |
| Qubit 1 | -1 | +1 | (1,0) |
| Qubit 2 | -1 | -1 | (1,1) |
| Qubit 3 | +1 | -1 | (0,1) |

#### Recovery

Based on syndrome $(s_1, s_2)$, apply correction:
- $(0,0)$ → No error; apply $I$.
- $(1,0)$ → Error on qubit 1; apply $\hat{\sigma}_x^{(1)}$.
- $(1,1)$ → Error on qubit 2; apply $\hat{\sigma}_x^{(2)}$.
- $(0,1)$ → Error on qubit 3; apply $\hat{\sigma}_x^{(3)}$.

**Result**: Original logical state $|\psi\rangle_L$ is recovered.

#### Key Insight

- **Syndrome measurement is non-destructive**: Measures parity (even/odd), not individual qubit values.
- **Doesn't collapse logical data**: Superposition $\alpha|000\rangle + \beta|111\rangle$ is preserved; only error type is revealed.
- **Protects against single errors**: Works if at most one qubit flips.

### General Quantum Error Correction

#### Requirements for a Code

1. **Distance** $d$: Code can correct up to $\lfloor (d-1)/2 \rfloor$ errors.
   - 3-qubit code: distance $d=3$, corrects $\lfloor 1 \rfloor = 1$ error.
   - [[7,1,3]] Hamming code: 7 physical qubits, 1 logical, distance 3; corrects 1 error.
   - [[5,1,3]] perfect code: 5 physical, 1 logical, distance 3; optimal.

2. **Stabilizer generators**: Independent parity checks that identify error type.
   - 3-qubit code: 2 independent stabilizers ($P_1, P_2$); specify error.

3. **Encoding efficiency**: Logical qubits vs. physical qubits.
   - 3-qubit: 1 logical per 3 physical; overhead factor 3.
   - Practical codes: overhead $\sim 1000$ physical qubits per logical (current technology).

#### Surface Code (Outlook)

2D array of qubits with local interactions; distance scales with physical system size. Leading candidate for fault-tolerant quantum computing.

### Quantum Key Distribution (BB84)

**Goal**: Alice and Bob securely share a random key that eavesdropper Eve cannot access.

#### Protocol

**Step 1**: Alice prepares random qubits in random bases:
- Bases: Rectilinear ($\{|0\rangle, |1\rangle\}$) and Diagonal ($\{|+\rangle, |-\rangle\}$).
- Bits: 0 → $|0\rangle$ or $|+\rangle$; 1 → $|1\rangle$ or $|-\rangle$ (depending on basis).

**Step 2**: Bob receives qubits; measures in random bases (doesn't know Alice's basis).

**Step 3**: Public channel: Alice and Bob openly announce bases (not values).

**Step 4**: Keep only bits where Alice and Bob used same basis; discard mismatched.

**Result**: Shared secret key (50% efficiency, but secure).

#### Security Against Eavesdropping

If Eve intercepts qubits:
- Eve measures in random basis (wrong with ~50% probability).
- Wrong measurement collapses qubit to wrong basis; Bob receives incorrect value.
- Alice and Bob detect **errors** exceeding expected noise level.
- Conclude eavesdropping; abort key generation.

**Unconditional security**: No computational complexity assumption; security guaranteed by quantum mechanics.

### Quantum Darwinism

**Concept** (Zurek): Environment gradually learns about a quantum system via interactions; information spreads to many environmental degrees of freedom.

#### Mechanism

1. System in superposition (e.g., $|{\uparrow}\rangle + |{\downarrow}\rangle$) interacts with environment.
2. Each environmental subsystem becomes entangled, encoding which state the system is in.
3. **Multiple copies** of state information exist in environment (redundancy).
4. Observer can learn about system without directly accessing it; observing small environmental fragment reveals system state.
5. Superposition is effectively "**selected out**" by environment; classicality emerges.

#### Objective Reality

**Definition**: Observable property is objective if multiple independent observers measure the same value.

**Darwinism explanation**: Environment stores redundant information about system; multiple observers (sampling different environmental fragments) learn the same system state. This creates **classical objectivity**.

**Example**: Pointer states (einselection) become classical-like; superpositions become classical ignorance (we don't know which state it's in, but it's in a definite state).

### No-Communication Theorem (Review)

Even with quantum entanglement (Bell pairs), one cannot send information faster than light:
- Teleportation requires 2 classical bits (light-speed limited).
- BB84 requires classical announcement of bases (light-speed limited).
- No way to send a message via entanglement alone without classical communication.

:::{admonition} No-cloning theorem
:class: important
An unknown quantum state cannot be perfectly copied. This is a direct consequence of the linearity of quantum mechanics and is the foundation of quantum cryptography: any eavesdropper attempting to copy a quantum key inevitably disturbs it.
:::

### Summary

- **No-cloning theorem**: Unknown quantum states cannot be universally copied; prevents simple error correction but enables unconditional security (BB84).
- **Quantum teleportation**: Send unknown state using Bell pair + 2 classical bits; requires pre-shared entanglement; no faster-than-light signaling.
- **Quantum error correction**: Encode logical qubit in multiple physical qubits; syndrome measurement identifies error type without collapsing data; 3-qubit code corrects 1 bit-flip error.
- **Requirements**: Distance $d$ (corrects $(d-1)/2$ errors), stabilizer generators (parity checks), encoding efficiency (overhead).
- **BB84 quantum key distribution**: Generate secure key by measuring qubits in random bases; eavesdropper detected by excess errors; unconditional security vs. computational security.
- **Quantum Darwinism**: Environment proliferates information about system in multiple copies; enables multiple observers to independently determine system state; explains emergence of classicality via environmental selection (einselection).
- **No-communication**: Despite quantum nonlocality, cannot transmit information faster than light; all practical protocols require classical channels.
