# 4.3 Landau Level

### Overview

Magnetic fields quantize orbital motion into discrete energy levels called Landau levels. These form the foundation of the quantum Hall effect, where Hall conductance is exactly quantized to parts per billion precision.

The IQHE is one of the most precisely measured phenomena in physics, with precision better than 1 part in $10^9$. It demonstrates that quantum mechanics produces perfectly quantized collective behavior at macroscopic scales, independent of material imperfections.

The quantum Hall effect reveals deep topological principles: Hall conductance quantization is a topological consequence (Chern number), not a band structure detail.

:::{admonition} Gauge Invariance in Magnetic Fields
:class: important

Landau levels are gauge-independent—the number of levels and their degeneracy do not depend on which gauge we choose for the vector potential. Yet the wavefunction structure changes dramatically between gauges.
:::
