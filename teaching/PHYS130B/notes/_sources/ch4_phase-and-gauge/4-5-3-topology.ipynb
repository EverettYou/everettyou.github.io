## Lecture Notes

The Berry curvature—the gauge field in parameter space—is a topological object: it can be integrated over closed surfaces to yield an integer, the Chern number, that cannot change without a phase transition. This topological invariant classifies quantum states according to their geometric structure and has profound consequences: the integer quantum Hall effect, topological insulators, and the stability of protected boundary states. Topology is not just mathematical elegance; it is a fundamental principle that determines what quantum states can exist.
:::{admonition} Chern Number Classifies Geometry
:class: important

The Chern number is a topological integer that counts how many times the Berry curvature "winds" around parameter space. It is a topological invariant—unchanged by smooth deformations of the parameter space.
:::


### Chern Number as Topological Invariant

The **Chern number** is defined for a band in a 2D crystal as:

$$

C_n = \frac{1}{2\pi}\int_{\text{BZ}}d^2k\\, \Omega_n(\boldsymbol{k})

$$

where the integral is over the entire Brillouin zone (BZ), which is a 2D torus topologically.

**Key properties**:

1. **Integer-valued**: By topological arguments, $C_n$ is always an integer: $C_n \in \mathbb{Z}$

2. **Topology-dependent**: The Chern number cannot change unless the band gap closes (system passes through a phase transition)

3. **Robust**: Small deformations of the band structure preserve $C_n$; disorder doesn't change it

4. **Gauge-invariant**: Although the Berry connection $\boldsymbol{A}_n$ is gauge-dependent, the curvature $\Omega_n$ and hence $C_n$ are gauge-invariant

### Proof of Integer Quantization

By applying Stokes' theorem and exploiting periodic boundary conditions on the torus-shaped Brillouin zone, the total Berry curvature integrated over the BZ must be a multiple of $2\pi$:

$$

\int_{\text{BZ}}\Omega = 2\pi n

$$

where $n$ is an integer. Therefore, the Chern number must be integer-valued:

$$

C_n = \frac{1}{2\pi}\int_{\text{BZ}}\Omega \in \mathbb{Z}

$$

This integrality is a topological consequence of the periodic boundary conditions (the BZ is a torus), not a coincidence. The Chern number cannot change continuously—it can only jump discontinuously when the band gap closes.


:::{admonition} Derivation
:class: dropdown note


To understand why $C_n$ must be an integer, apply Stokes' theorem to a small region of the BZ:

$$

\int_S \Omega \cdot d\boldsymbol{A} = \oint_C \boldsymbol{A} \cdot d\boldsymbol{k}

$$

When integrating over the entire closed BZ (which is topologically a torus), we can decompose it into small surface patches. The line integrals on shared boundaries cancel, leaving:

$$

\int_{\text{BZ}}\Omega = \text{sum of boundary line integrals}

$$

Due to periodic boundary conditions on the torus, the wavefunction phase must return to itself after traversing the BZ. This quantizes the total flux:

$$

\int_{\text{BZ}}\Omega = 2\pi n

$$

with integer $n$. Thus:

$$

C_n = \frac{1}{2\pi}\int_{\text{BZ}}\Omega = n \in \mathbb{Z}

$$

:::
### Winding Number and Berry Phase

For a 1D system, the **winding number** plays the role of the Chern number. As momentum $k$ goes from 0 to $2\pi$ (around the 1D Brillouin zone), the phase of the wavefunction winds around the complex plane some integer number of times.

The **Berry phase for a closed 1D path** is:

$$

\gamma = i\oint_0^{2\pi}\langle\psi(k)|\frac{d}{dk}\psi(k)\rangle dk

$$

This can be written as:

$$

\gamma = \arg[\langle\psi(2\pi)|\psi(0)\rangle] = 2\pi w

$$

where $w$ is the **winding number**—an integer counting how many times the phase wraps around.

### Fiber Bundle Picture

To understand Berry phase's geometric meaning, we use the language of **fiber bundles** from differential geometry:

**Base space**: The parameter space (configuration space, momentum space, or nuclear coordinates). Example: the 2D Brillouin zone for band theory.

**Fiber**: At each point in parameter space, we have a Hilbert space (or subspace for a single band). The fiber is the space of quantum states.

**Section**: A choice of eigenstate $|n(\boldsymbol{R})\rangle$ as a function of parameters—a "section" of the bundle.

**Connection**: The Berry connection $\boldsymbol{A}_n = i\langle n|\nabla n\rangle$ is the **connection 1-form** that specifies how to parallel-transport quantum states through parameter space.

**Curvature**: The Berry curvature $\Omega = d\boldsymbol{A} + \boldsymbol{A} \wedge \boldsymbol{A}$ (for non-abelian bundles) characterizes the intrinsic curvature of the bundle.

This geometric language reveals that Berry phase is fundamentally a **connection on a fiber bundle**—the same mathematical structure that underlies gauge theories and gravitation.

### Topological Insulators and Bulk-Boundary Correspondence

A **topological insulator** is a material that is insulating in the bulk but has conducting (or perfectly transmitted) edge states. The protection mechanism involves Berry curvature and Chern numbers.

**2D Topological Insulator**: (Also called a **Chern insulator**)

- The occupied bands have non-zero total Chern number: $C = \sum_n C_n \neq 0$
- The bulk is gapped (insulating): energy gap between occupied and unoccupied bands
- **Protected edge states**: The boundary has exactly $|C|$ chiral edge states that cross the band gap
- These edge states conduct current even if the bulk is a perfect insulator

**3D Topological Insulator**: (Z_2 topological insulator)

- More subtle: cannot be characterized by a single Chern number
- Involves a $\mathbb{Z}_2$ topological invariant (integers modulo 2)
- Protected surface states exist on the 2D boundary
- Examples: Bi₂Se₃, Bi₂Te₃

### Bulk-Boundary Correspondence Theorem

A deep topological principle: **the number of edge states equals the Chern number of bulk bands**.

**Why?** When we cut the infinite 2D system to create a boundary:

1. Bulk eigenstates remain (with gaps between occupied and unoccupied bands)
2. New states appear at the edge with energies in the bulk gap
3. These edge states must connect occupied and unoccupied bands as we traverse the boundary
4. Topological argument: the number of such edge states is pinned to equal the bulk Chern number

This is **topological protection**—the edge states cannot be eliminated by perturbations without closing the gap. Disorder, impurities, or deformations at the edge don't destroy the edge current because the edge states form a complete 1D band that must be filled.

### Integer Quantum Hall Effect from Topology

The integer quantum Hall effect is understood completely through Chern number topology:

1. When $\nu$ Landau levels are filled, the occupied bands have total Chern number $C = \nu$

2. By bulk-boundary correspondence, there are $\nu$ chiral edge states

3. Each edge state carries conductance $e^2/h$, giving total Hall conductance:

$$

\sigma_H = \nu \times \frac{e^2}{h}

$$

4. The quantization is **exact and universal** because it's a topological property—independent of disorder, material details, or microscopic parameters

### Topological Phase Transitions

When a system undergoes a topological phase transition, the Chern number changes. This can only happen when **the band gap closes**—states from occupied and unoccupied bands become degenerate.

At a critical parameter value:
- The band gap vanishes
- States from different Chern number sectors "mix"
- Beyond the transition, the new topological order emerges

This is profoundly different from conventional phase transitions (like ferromagnet to paramagnet), which are characterized by symmetry breaking. Topological transitions involve no symmetry change—only a change in global topology.

### Summary

- **Chern number**: Integer topological invariant $C = \frac{1}{2\pi}\int\Omega d^2k$ insensitive to smooth deformations.
- **Quantization**: Periodic boundary conditions on torus (BZ) force $C \in \mathbb{Z}$.
- **Winding number**: 1D analog; counts phase windings as momentum traverses Brillouin zone.
- **Fiber bundle geometry**: Berry phase is connection on bundle of quantum states over parameter space.
- **Bulk-boundary correspondence**: Number of edge states = Chern number of bulk bands.
- **Topological insulator**: Bulk gap + non-zero Chern number → protected conducting edge states.
- **IQHE from topology**: $\sigma_H = \nu e^2/h$ follows from bulk Chern number $\nu$.
- **Robustness**: Topological properties protected against disorder and impurities (as long as gap survives).
- **Topological phase transition**: Band gap closure allows Chern number to change.
- **Deep principle**: Quantum geometry (Berry phase, curvature) directly determines observable transport properties.
