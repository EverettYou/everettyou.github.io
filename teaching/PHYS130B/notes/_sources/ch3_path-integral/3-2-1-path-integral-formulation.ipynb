## Lecture Notes

Feynman's path integral reveals that a quantum particle does not follow a single trajectory—it explores every possible path simultaneously, with each path contributing a phase proportional to its classical action. This radical idea unifies quantum mechanics, statistical mechanics, and quantum field theory into a single mathematical framework.

### Feynman's Path Integral Principle

In 1948, **Richard Feynman** proposed a radically different formulation of quantum mechanics based on a deceptively simple idea:

:::{admonition} Path Integral Principle
:class: important

The quantum amplitude for a particle to evolve from an initial state at $(\boldsymbol{x}_i, t_i)$ to a final state at $(\boldsymbol{x}_f, t_f)$ is the **sum over all possible paths** $\boldsymbol{x}(t)$ connecting these points:

$$

\psi(\boldsymbol{x}_f, t_f) = \int_{\boldsymbol{x}(t_i)=\boldsymbol{x}_i}^{\boldsymbol{x}(t_f)=\boldsymbol{x}_f} e^{iS[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}] \, \psi(\boldsymbol{x}_i, t_i)

$$ (eq-feynman-path-integral)

where:
- $S[\boldsymbol{x}] = \int_{t_i}^{t_f} L(\boldsymbol{x}, \dot{\boldsymbol{x}}, t) \, dt$ is the **action functional** along path $\boldsymbol{x}(t)$.
- $\mathcal{D}[\boldsymbol{x}]$ denotes a **functional integral** (sum) over all paths.
- Each path contributes with **magnitude** 1 (unit weight) and **phase** $e^{iS/\hbar}$.
:::

This is a profound reformulation. Rather than solving a differential equation (Schrödinger equation), we sum over all possibilities. The quantum evolution is a **sum of probability amplitudes**, not a unique deterministic trajectory.

### The Propagator: Transition Amplitude

The **propagator** (also called **Green's function** or **transition amplitude**) is

$$

K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int_{\boldsymbol{x}(t_i)=\boldsymbol{x}_i}^{\boldsymbol{x}(t_f)=\boldsymbol{x}_f} e^{iS[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$ (eq-propagator-definition)

It represents the amplitude for a particle at $\boldsymbol{x}_i$ at time $t_i$ to arrive at $\boldsymbol{x}_f$ at time $t_f$. 

**Wavefunction evolution**: Given an initial wavefunction $\psi(\boldsymbol{x}_i, t_i)$, the wavefunction at a later time is

$$

\psi(\boldsymbol{x}_f, t_f) = \int K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) \, \psi(\boldsymbol{x}_i, t_i) \, d^3\boldsymbol{x}_i

$$ (eq-wavefunction-evolution)

The propagator is the **kernel** for time evolution.

#### Composition Property (Chain Rule for Propagators)

Suppose we want to go from time $t_i$ to time $t_f$, but we know the state at an intermediate time $t_m$ (with $t_i < t_m < t_f$). Then

$$

K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_m, t_m) K(\boldsymbol{x}_m, t_m; \boldsymbol{x}_i, t_i) \, d^3\boldsymbol{x}_m

$$ (eq-propagator-composition)

**Physical interpretation**: A path from $t_i$ to $t_f$ can be decomposed into a path from $t_i$ to $t_m$ and then from $t_m$ to $t_f$. Integrating over all possible intermediate positions at $t_m$ gives the full propagator.

**Proof**: Consider the path integral from $t_i$ to $t_f$:

$$

K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int_{\boldsymbol{x}(t_i)=\boldsymbol{x}_i}^{\boldsymbol{x}(t_f)=\boldsymbol{x}_f} e^{iS[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$

Any path from $t_i$ to $t_f$ passes through some point $\boldsymbol{x}_m$ at time $t_m$. Grouping paths by their value at $t_m$:

$$

K = \int d^3\boldsymbol{x}_m \left( \int_{t_i \to t_m} e^{iS[t_i \to t_m]/\hbar} \, \mathcal{D}[\text{paths } t_i \to t_m] \right) \left( \int_{t_m \to t_f} e^{iS[t_m \to t_f]/\hbar} \, \mathcal{D}[\text{paths } t_m \to t_f] \right)

$$

Since the action is additive, $S[t_i \to t_f] = S[t_i \to t_m] + S[t_m \to t_f]$, the exponentials factor:

$$

e^{iS[t_i \to t_f]/\hbar} = e^{iS[t_i \to t_m]/\hbar} \cdot e^{iS[t_m \to t_f]/\hbar}

$$

This gives the composition formula.

### The Action: Lagrangian Path Integral

The **action** along a path $\boldsymbol{x}(t)$ is defined as

$$

S[\boldsymbol{x}] = \int_{t_i}^{t_f} L(\boldsymbol{x}(t), \dot{\boldsymbol{x}}(t), t) \, dt

$$ (eq-action-definition)

where $L$ is the **Lagrangian**. For a non-relativistic particle of mass $m$ in a potential $V$:

$$

L = \frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x}, t)

$$ (eq-lagrangian-standard)

The action has dimensions of [energy] $\times$ [time] = [angular momentum], as does $\hbar$.

#### Why the Lagrangian?

The Lagrangian is fundamental in classical mechanics because the **principle of stationary action** states:

$$

\delta S = 0 \quad \Rightarrow \quad \text{Equations of motion}

$$ (eq-stationary-action-principle)

The path followed by a classical particle is the one for which $S$ is stationary (extremized). Varying $S$ with respect to the path gives the Euler-Lagrange equations:

$$

\frac{d}{dt}\left(\frac{\partial L}{\partial \dot{x}^i}\right) - \frac{\partial L}{\partial x^i} = 0

$$ (eq-euler-lagrange)

In quantum mechanics, the path integral **retains all paths**, but each is weighted by the classical action. This is a beautiful synthesis: quantum mechanics respects the classical action, but explores all possibilities.

### Physical Interpretation: Amplitude and Interference

#### Equal Magnitude Contribution

Each path contributes with **unit amplitude** (magnitude 1): $|e^{iS/\hbar}| = 1$. This may seem counterintuitive—shouldn't "unlikely" paths contribute less? In the path integral framework, all paths contribute equally in amplitude; what matters is their **relative phase**.

#### Phase Weighting: Interference

Two paths $\boldsymbol{x}_1(t)$ and $\boldsymbol{x}_2(t)$ have phases $\Phi_1 = S_1/\hbar$ and $\Phi_2 = S_2/\hbar$. Their **phase difference** is

$$

\Delta\Phi = \frac{S_1 - S_2}{\hbar}

$$ (eq-phase-difference)

**Constructive interference**: When $\Delta\Phi = 2\pi n$ (phase difference is a multiple of $2\pi$), the two paths interfere constructively—their amplitudes add.

**Destructive interference**: When $\Delta\Phi = \pi(2n+1)$ (phase difference is an odd multiple of $\pi$), the paths interfere destructively—their amplitudes partially cancel.

**Stationary action paths**: If two paths have nearly the same action (differ by $\lesssim \hbar$), their phases are nearly equal, and they interfere constructively. Paths clustered around a stationary point of $S$ all have similar phases and add coherently.

#### Classical Path Dominance

The **classical path** is defined by $\delta S = 0$. Near the classical path:

$$

S[\boldsymbol{x}_\text{cl} + \delta\boldsymbol{x}] \approx S[\boldsymbol{x}_\text{cl}] + \frac{1}{2}(\delta\boldsymbol{x})^T M (\delta\boldsymbol{x}) + \cdots

$$

where $M$ is the second variation (Hessian). For a small deviation $\delta\boldsymbol{x} \sim \sqrt{\hbar}$, the action difference is $\Delta S \sim \hbar$, giving a phase difference $\sim 1$ radian.

**In the classical limit ($\hbar \to 0$)**:
- Paths with $|S - S_\text{cl}| \gg \hbar$ have rapidly oscillating phases and **cancel destructively**.
- Only paths with $|S - S_\text{cl}| \lesssim \hbar$ (a neighborhood of the classical path) **interfere constructively**.
- As $\hbar \to 0$, the constructive region shrinks to zero, and only the classical path survives.

:::{admonition} Quantum-to-classical correspondence
:class: note

The path integral naturally explains the correspondence principle:

- **Macroscopic systems**: $S \gg \hbar$ for every path in the problem. The classical path's action dominates. Quantum fluctuations are negligible.

- **Microscopic systems**: $S \sim \hbar$. Many paths contribute significantly. Quantum effects are important.

- **Classical limit**: $\hbar \to 0$ (or equivalently, $S \to \infty$ relative to $\hbar$). The classical path is absolutely selected. Quantum mechanics reduces to classical mechanics.
:::

### Path Integral vs. Schrödinger Equation

The path integral and Schrödinger equation are **equivalent** formulations of quantum mechanics, but they provide different insights.

| Aspect | Schrödinger Equation | Path Integral |
|--------|---------------------|---------------|
| Formulation | Differential equation (PDE) | Functional integral |
| Variables | Wavefunction $\psi(\boldsymbol{x}, t)$ | All paths $\boldsymbol{x}(t)$ |
| Causality | Forward in time (temporal boundary condition) | All paths (spacetime picture) |
| Classical limit | Implicit (via stationary phase) | Explicit (stationary action, $\hbar \to 0$) |
| Computational | Direct for simple problems; perturbation theory | Systematic perturbation and approximations |
| Physical intuition | Wave-particle duality | Sum of probability amplitudes |

**Derivability**: The Schrödinger equation can be derived from the path integral using the composition property and a small-time-step approximation. Conversely, the path integral can be constructed as a solution to the Schrödinger equation using the propagator.

### Free Particle as Motivation

For a **free particle** (no potential, $V = 0$), the Lagrangian is $L = \frac{m}{2}\dot{\boldsymbol{x}}^2$. The action is

$$

S[\boldsymbol{x}] = \int_{t_i}^{t_f} \frac{m}{2}\dot{\boldsymbol{x}}^2 \, dt

$$ (eq-free-action)

Classically, the extremum is a **straight line** (constant velocity):

$$

\boldsymbol{x}_\text{cl}(t) = \boldsymbol{x}_i + \frac{\boldsymbol{x}_f - \boldsymbol{x}_i}{t_f - t_i}(t - t_i)

$$ (eq-classical-trajectory)

with action

$$

S_\text{cl} = \frac{m}{2}\left(\frac{\boldsymbol{x}_f - \boldsymbol{x}_i}{t_f - t_i}\right)^2 (t_f - t_i) = \frac{m(\boldsymbol{x}_f - \boldsymbol{x}_i)^2}{2(t_f - t_i)}

$$ (eq-free-action-classical)

For a free particle, the path integral can be **evaluated exactly** (Section 3.2.2), giving a Gaussian result.

### Summary

- **Path integral**: Quantum amplitude is a sum over all paths, each weighted by $e^{iS/\hbar}$.

- **Propagator**: Transition amplitude between two spacetime points; solves the Schrödinger equation.

- **Composition property**: Propagators multiply for sequential time intervals.

- **Action functional**: $S[\boldsymbol{x}] = \int L \, dt$ is the classical action. Stationary points give classical equations of motion.

- **Interference**: Paths with nearly equal action interfere constructively; paths with differing action cancel.

- **Classical limit**: $\hbar \to 0$ suppresses paths away from stationary action; classical mechanics emerges.

- **Equivalence**: Path integral and Schrödinger equation are equivalent formulations of quantum mechanics.


:::{admonition} Derivation: Gaussian Path Integral for a Single Time Slice
:class: dropdown note

For a free particle propagating over a time interval $\varepsilon$, the action is

$$S[\boldsymbol{x}] = \int_0^\varepsilon \frac{m}{2}\dot{\boldsymbol{x}}^2 \, dt

$$

For a path from $\boldsymbol{x}_i$ to $\boldsymbol{x}_f$ in time $\varepsilon$, we parameterize the path and evaluate the path integral.

In the **time-sliced formulation**, the path integral becomes

$$K(\boldsymbol{x}_f, \boldsymbol{x}_i; \varepsilon) = \int \mathcal{D}[\boldsymbol{x}] \, e^{iS[\boldsymbol{x}]/\hbar}

$$

For small $\varepsilon$, the path is approximately a straight line, and we can write

$$\boldsymbol{x}(t) = \boldsymbol{x}_i + \frac{(\boldsymbol{x}_f - \boldsymbol{x}_i)\varepsilon}{\varepsilon} t = \boldsymbol{x}_i + (\boldsymbol{x}_f - \boldsymbol{x}_i)\frac{t}{\varepsilon}

$$

This gives

$$S = \frac{m}{2} \left(\frac{\boldsymbol{x}_f - \boldsymbol{x}_i}{\varepsilon}\right)^2 \varepsilon = \frac{m(\boldsymbol{x}_f - \boldsymbol{x}_i)^2}{2\varepsilon}

$$

The path integral is a **Gaussian integral**:

$$K = \int \mathcal{D}[\boldsymbol{x}] \, \exp\left(\frac{im(\boldsymbol{x}_f - \boldsymbol{x}_i)^2}{2\hbar\varepsilon}\right)

$$

Using the Gaussian integral formula $\int e^{-\alpha x^2} dx = \sqrt{\pi/\alpha}$ (with $\alpha \to -i\alpha$):

$$K(\boldsymbol{x}_f, \boldsymbol{x}_i; \varepsilon) = \left(\frac{m}{2\pi i\hbar\varepsilon}\right)^{d/2} \exp\left(\frac{im(\boldsymbol{x}_f - \boldsymbol{x}_i)^2}{2\hbar\varepsilon}\right)

$$

where $d$ is the spatial dimension. The prefactor $(m/2\pi i\hbar\varepsilon)^{d/2}$ is the **normalization factor** for a single time slice.

:::


:::{admonition} Derivation: Normalization Factor for the Path Integral Measure
:class: dropdown note

The path integral measure $\mathcal{D}[\boldsymbol{x}]$ for $N$ discrete time slices is defined as

$$\mathcal{D}[\boldsymbol{x}] = \prod_{n=1}^{N-1} d^d\boldsymbol{x}_n \cdot \lambda_N^{-N/2}

$$

where $\lambda_N$ is a **normalization constant** (dimension [length]$^d$) that ensures proper dimensions and convergence.

For a free particle, each time slice contributes a factor

$$\lambda_\varepsilon = \left(\frac{2\pi i\hbar\varepsilon}{m}\right)

$$

For $N$ slices spanning total time $T = N\varepsilon$:

$$\lambda_N = \lambda_\varepsilon = \frac{2\pi i\hbar\varepsilon}{m}

$$

The full propagator over $N$ slices is

$$K_N = \prod_{n=0}^{N-1} \left(\frac{m}{2\pi i\hbar\varepsilon}\right)^{d/2} \exp\left(\frac{im(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2}{2\hbar\varepsilon}\right)

$$

$$= \left(\frac{m}{2\pi i\hbar\varepsilon}\right)^{Nd/2} \int \prod_{n=1}^{N-1} d^d\boldsymbol{x}_n \, \exp\left(\frac{i}{\hbar} \sum_n \frac{m}{2}\frac{(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2}{\varepsilon}\right)

$$

The prefactor $\left(m/2\pi i\hbar\varepsilon\right)^{Nd/2}$ is the product of normalization factors for each time slice, ensuring that probability is conserved: $\int K_N(\boldsymbol{x}_f, \boldsymbol{x}_i) \, d^d\boldsymbol{x}_i = 1$ for fixed $\boldsymbol{x}_f$.

:::
