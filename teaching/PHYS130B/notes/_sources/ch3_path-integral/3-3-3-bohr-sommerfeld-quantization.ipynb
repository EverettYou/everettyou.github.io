## Lecture Notes

:::{admonition} Definition
:class: important

**Bohr-Sommerfeld quantization** reveals that discrete energy levels arise from a simple geometric condition: the phase accumulated around a closed classical orbit must equal an integer multiple of $2\pi$. This bridges classical mechanics and quantum mechanics, explaining quantization without solving the Schrödinger equation exactly.
:::

### Quantization from Single-Valuedness

In quantum mechanics, the wavefunction must be **single-valued** in configuration space. For a bound state (particle oscillating in a potential well), the wavefunction returns to itself after a complete cycle. This imposes a **quantization condition** on the allowed energies.

#### The Closed Orbit Condition

For a **bound state** in a 1D potential, the particle oscillates between classical turning points $x_a$ and $x_b$ (where $E = V(x)$). The phase accumulated on the left side of the barrier (from $x_a$ to $x_b$) is

$$\Phi_R = \frac{1}{\hbar}\int_{x_a}^{x_b} p(x) dx$$


By symmetry, the phase on the return (from $x_b$ to $x_a$) is the same. The **total phase** for a complete cycle is

$$\Phi_\text{total} = \frac{2}{\hbar}\int_{x_a}^{x_b} p(x) dx + \frac{\pi}{2}$$


At a **simple turning point**, the WKB connection formula introduces a phase shift of $\pi/4$ (or $\pi/2$ total, counting both turning points).

:::{admonition} Discussion: Semi-Classical Quantization
:class: dropdown warning

Bohr-Sommerfeld quantization condition $\oint p \, \mathrm{d}x = 2\pi\hbar(n + 1/2)$ emerges from WKB by requiring single-valued wavefunction around a closed orbit. Yet it looks almost *classical* (an integral around a closed path). This raises foundational questions:

- Bohr-Sommerfeld correctly predicts energy levels for hydrogen (like the quantum result for $n$ levels). For more complex atoms, it fails. Why does it work for hydrogen but not for helium?
- The factor $1/2$ (zero-point energy) comes from phase shifts at turning points. Is this factor purely quantum, or does it have a semi-classical interpretation?
- For a chaotic (non-integrable) system, the Bohr-Sommerfeld condition doesn't apply (no longer valid action variables). Yet quantum mechanics still works. How does quantum mechanics handle chaos when semi-classical intuition fails?
- Can you design a potential where Bohr-Sommerfeld gives an *exact* result (matching the full quantum spectrum exactly)? What special property must the potential have?
:::
#### Single-Valuedness Requirement

For the wavefunction to be single-valued after a complete oscillation, the total phase must be a multiple of $2\pi$:

$$\Phi_\text{total} = 2\pi n, \quad n = 0, 1, 2, \ldots$$


Therefore:

$$\frac{2}{\hbar}\int_{x_a}^{x_b} p(x) dx + \frac{\pi}{2} = 2\pi n$$


Simplifying:

$$\oint p(x) dx = 2\pi\hbar\left(n + \frac{1}{2}\right), \quad n = 0, 1, 2, \ldots$$


:::{admonition} The Bohr-Sommerfeld Rule
:class: important

For a 1D bound state, the allowed energies $E_n$ satisfy

$$\oint p(x) dx = 2\pi\hbar\left(n + \frac{1}{2}\right)$$


where $p(x) = \sqrt{2m(E - V(x))}$ and the integral is around the complete classical orbit. The factor of $1/2$ is the **zero-point energy correction**—even the ground state ($n=0$) has non-zero energy.
:::

:::{admonition} Example: Harmonic Oscillator
:class: example

For the harmonic oscillator $V(x) = \frac{1}{2}m\omega^2 x^2$:

**Turning points**: At $E = V(x)$, we have $x = \pm a$ where $a = \sqrt{2E/(m\omega^2)}$.

**Momentum**: $p(x) = m\omega\sqrt{a^2 - x^2}$.

**Integral**: 

$$\oint p dx = 2\int_{-a}^{a} m\omega\sqrt{a^2 - x^2} dx = \pi m\omega a^2 = \frac{2\pi E}{\omega}$$


**Quantization condition**:

$$\frac{2\pi E_n}{\omega} = 2\pi\hbar\left(n + \frac{1}{2}\right)$$


$$E_n = \hbar\omega\left(n + \frac{1}{2}\right), \quad n = 0, 1, 2, \ldots$$


This is **exactly the quantum mechanical result**! For the harmonic oscillator, Bohr-Sommerfeld is exact.
:::

:::{admonition} Note: Why is the Harmonic Oscillator Special?
:class: note

The harmonic oscillator is special because its potential is quadratic. In the WKB approximation, the momentum $p(x) \propto \sqrt{E - V}$ leads to an integral that is tractable exactly. Moreover, the WKB phase shifts are exact for quadratic potentials. For higher-order potentials, WKB and Bohr-Sommerfeld are excellent approximations but not exact.
:::

:::{admonition} Example: Hydrogen Atom (Bohr Model)
:class: example

For the Coulomb potential $V(r) = -k/r$ with circular orbits, angular momentum quantization $L = n_\phi\hbar$ combined with radial quantization yields

$$E_n = -\frac{13.6 \text{ eV}}{n^2}, \quad n = 1, 2, 3, \ldots$$


This agrees with the quantum mechanical result! However, Bohr-Sommerfeld requires careful treatment of the radial turning point and angular momentum quantization separately. The result shows that Bohr-Sommerfeld works even for non-separable or complicated potentials when applied systematically.
:::

### EBK Quantization: Integrable Systems

The **Einstein-Brillouin-Keller (EBK) quantization** generalizes Bohr-Sommerfeld to systems with multiple degrees of freedom that are **integrable** (i.e., have as many constants of motion as degrees of freedom).

#### Multiple Degrees of Freedom

For a system with $N$ degrees of freedom and $N$ action variables $J_i$, the quantization condition is

$$J_i = \oint p_i dq_i = 2\pi\hbar(n_i + \alpha_i), \quad n_i = 0, 1, 2, \ldots$$


where $\alpha_i$ is the **Maslov index** (phase correction), typically $1/2$ for each degree of freedom with turning points.

:::{admonition} Example: 2D Isotropic Oscillator
:class: example

For the 2D harmonic oscillator $V(x, y) = \frac{1}{2}m\omega^2(x^2 + y^2)$, with action variables $J_x = \oint p_x dx$ and $J_y = \oint p_y dy$ and using EBK quantization $J_i = 2\pi\hbar(n_i + 1/2)$:

$$E_{n_x, n_y} = \hbar\omega(n_x + n_y + 1), \quad n_x, n_y = 0, 1, 2, \ldots$$


Again, exact agreement with quantum mechanics. EBK correctly captures the degeneracies and energy level structure.
:::

### Accuracy and Limitations

#### When Bohr-Sommerfeld Works Well

1. **Quadratic or near-quadratic potentials**: Harmonic oscillators and slowly-varying deformations
2. **Deep potential wells**: Large quantum numbers $n \gg 1$ (large $S/\hbar$)
3. **Separable systems**: Multiple uncoupled degrees of freedom (use EBK)
4. **WKB-valid potentials**: Smooth, slowly-varying (potential derivatives small on wavelength scale)

#### When Bohr-Sommerfeld Fails

1. **Steep or discontinuous potentials**: Violates WKB assumptions
2. **Degeneracies**: Degenerate states require careful treatment
3. **Resonances**: Coupled oscillators with irrational frequency ratios
4. **Chaotic systems**: Non-integrable dynamics; no action variables

#### Accuracy for Hydrogen

For hydrogen, Bohr-Sommerfeld gives exact energy levels but incorrect radial wave functions. The reason is subtle: exact cancellation of errors in the action integral, combined with spherical symmetry. For slightly different potentials (e.g., helium atom), Bohr-Sommerfeld accuracy drops to $\sim 5\text{-}10\%$.

#### Improving Bohr-Sommerfeld: Higher-Order Corrections

Beyond the leading WKB order, systematic corrections can be computed. These go as powers of $1/n$ (for large quantum number $n$) and involve higher derivatives of the potential. For high-$n$ states, WKB corrections typically improve accuracy to $\sim 0.1\%$.

### Connection to Path Integral

The Bohr-Sommerfeld rule is a **direct consequence of the path integral and stationary phase approximation**. In the path integral for the propagator between turning points, only paths with stationary action survive. For a bound state, the action integral around the closed orbit encodes the quantization condition.

$$\text{Bohr-Sommerfeld} = \text{Stationary phase applied to periodic orbits}$$


This is why the path integral naturally bridges classical mechanics (action, orbits) and quantum mechanics (phase, quantization).

:::{admonition} Note: Classical Limit and Correspondence Principle
:class: note

In the classical limit ($n \gg 1$), the energy spacing $\Delta E = E_{n+1} - E_n = \hbar\omega$ should match the classical frequency via the correspondence principle. For the harmonic oscillator, this is exact. For general potentials with Bohr-Sommerfeld, the energy level spacing approaches the classical oscillation frequency, confirming the smooth transition to classical mechanics.
:::

### Summary

- **Single-valuedness requirement**: Wavefunction must return to itself after a complete orbit, quantizing the phase
- **Bohr-Sommerfeld rule**: $\oint p dx = 2\pi\hbar(n + 1/2)$ determines allowed energies without solving the Schrödinger equation
- **Zero-point energy**: The $1/2$ factor ensures non-zero ground-state energy, a quintessentially quantum effect
- **Harmonic oscillator**: Bohr-Sommerfeld is exact, giving $E_n = \hbar\omega(n + 1/2)$
- **Hydrogen atom**: Bohr-Sommerfeld gives correct energy levels when applied carefully (azimuthal + radial quantization)
- **EBK quantization**: Generalizes to integrable systems with multiple degrees of freedom
- **Validity and accuracy**: Excellent for high-$n$ states and smooth potentials; fails for chaos and discontinuities
- **Path integral origin**: Bohr-Sommerfeld is the stationary-phase result for periodic orbits in the path integral
- **Classical correspondence**: Energy level spacing matches classical frequency, confirming the semiclassical limit
