## Lecture Notes

Huygens' principle reveals that light propagation is fundamentally about superposition of wavelets, each carrying a phase proportional to optical path length. In the high-frequency limit, constructive interference from paths of stationary optical path length produces Fermat's principle—geometric optics emerges from wave physics.

:::{admonition} Discussion: Secondary Wavelets
:class: dropdown warning

Huygens principle says each wavefront point radiates secondary spherical wavelets. The new wavefront emerges from their superposition. Yet a subtle question arises: do all paths from the aperture to the observation point contribute equally?

- If every aperture element contributes a secondary wavelet with equal *amplitude*, most contributions would interfere destructively. Yet diffraction patterns show sharp maxima, not averaging. How does interference select certain paths over others?
- In single-slit diffraction, does the edge of the aperture (abrupt boundary) create a discontinuity in secondary wavelets? Can Huygens principle handle sharp edges rigorously, or must we smooth them?
- The stationary phase approximation later in this chapter says only paths near the classical (stationary) path dominate. Is this the same selection mechanism as Huygens' destructive interference, just viewed differently?
- For a small aperture (width comparable to wavelength), can you predict the diffraction pattern using only Huygens principle plus superposition? What happens if the aperture is much smaller than the wavelength?
:::

### Huygens' Principle: The Foundation of Wave Optics

**Christiaan Huygens** (1690) introduced a constructive principle for understanding wave propagation.

:::{admonition} Key Result: Huygens' Principle
:class: important

Each point on a wavefront acts as a source of **secondary wavelets**. These wavelets are spherical waves emanating from that point. The wavefront at a later time is the **envelope** (coherent superposition) of all these secondary wavelets.
:::

This principle is not derived from Maxwell's equations (which didn't exist in Huygens' time) but rather a geometric statement about how waves evolve. Yet it captures the essence of wave behavior.

#### Elementary Wavelet and Phase

Consider a point source on the wavefront at position $\boldsymbol{r}_0$ at time $t=0$. It emits a spherical wave that reaches position $\boldsymbol{r}$ at time $t$:

$$


\psi(\boldsymbol{r}, t) \propto \frac{\mathrm{e}^{\mathrm{i}(k|\boldsymbol{r}-\boldsymbol{r}_0| - \omega t)}}{|\boldsymbol{r}-\boldsymbol{r}_0|}

$$ (eq-spherical-wave)

The **phase** of this wavelet is

$$


\Phi(\boldsymbol{r}, t) = k|\boldsymbol{r}-\boldsymbol{r}_0| - \omega t = \frac{2\pi}{\lambda}|\boldsymbol{r}-\boldsymbol{r}_0| - 2\pi f t

$$ (eq-wavelet-phase)

where $\lambda = 2\pi/k$ is the wavelength and $f = \omega/(2\pi)$ is the frequency. The distance traveled is $|\boldsymbol{r}-\boldsymbol{r}_0|$, and the phase accumulated is proportional to this distance (in units of wavelength).

#### Inhomogeneous Media: Optical Path Length

When light propagates through a medium with varying refractive index $n(\boldsymbol{r})$, the **optical path length** from $\boldsymbol{r}_0$ to $\boldsymbol{r}$ is

$$


L(\boldsymbol{r}_0 \to \boldsymbol{r}) = \int_{\boldsymbol{r}_0}^{\boldsymbol{r}} n(\boldsymbol{r}') \, \mathrm{d}s'

$$ (eq-optical-path-length)

where $ds'$ is the arc length along the path. The phase accumulated is

$$


\Phi = k L(\boldsymbol{r}_0 \to \boldsymbol{r}) - \omega t = \frac{2\pi}{\lambda_0} L(\boldsymbol{r}_0 \to \boldsymbol{r}) - 2\pi f t

$$ (eq-optical-path-phase)

where $\lambda_0$ is the **vacuum wavelength**. **Crucial observation**: Huygens' principle weights each secondary wavelet by its phase, which depends on optical path length.

:::{admonition} Key Result: Huygens-Fresnel Integral
:class: important

The amplitude at an observation point $B$ is the superposition over all points on an intermediate wavefront or aperture:

$$


\psi_B = \frac{1}{\mathrm{i}\lambda} \int_S K(\boldsymbol{r}) \, \mathrm{e}^{\mathrm{i}kL(A \to \boldsymbol{r} \to B)} \, \mathrm{d}S

$$ (eq-huygens-fresnel)

where $S$ is the unobstructed aperture surface, $L$ is optical path length, and $K(\boldsymbol{r})$ is an obliquity factor. This is a **sum over paths weighted by phase** proportional to optical path—the precursor to Feynman's path integral.
:::

### Deriving Refraction from Huygens' Principle

Consider a plane wavefront in medium 1 (refractive index $n_1$) incident on a boundary with medium 2 (refractive index $n_2$). Different points on the wavefront reach the boundary at different times. Each point becomes a secondary source.

**Setup**:
- Point $A$ on the wavefront reaches the boundary at time $t=0$.
- Point $B$ on the same wavefront, a distance $\ell$ away, reaches the boundary at time $t_B = \ell/(v_1) = \ell n_1/c$ (where $v_1 = c/n_1$).

**Secondary wavelets**:
- At point $A$: wavelet propagates a distance $d_A$ into medium 2 in time $t_B$, so $d_A = (c/n_2) t_B = \ell n_1/n_2$.
- At point $B$: wavelet has just started propagating, so $d_B = 0$.

**New wavefront envelope**: The envelope of secondary wavelets is a plane at angle $\theta_2$ to the normal, where

$$


\sin\theta_2 = \frac{\mathrm{d}_A}{\ell} = \frac{n_1}{n_2} \sin\theta_1

$$ (eq-snell-law)

This is **Snell's law of refraction**, derived purely from geometry and Huygens' principle without needing Maxwell's equations.

:::{admonition} Derivation: Snell's Law from Fermat's Principle
:class: dropdown note

Fermat's principle states that light follows the path for which the optical path length is stationary (extremal). Given a path from point $A$ in medium 1 (refractive index $n_1$) to point $B$ in medium 2 (refractive index $n_2$), crossing the interface at position $x$, the optical path length is

$$L(x) = n_1\sqrt{d_1^2 + x^2} + n_2\sqrt{d_2^2 + (w-x)^2}$$


where $d_1$, $d_2$ are the perpendicular distances from $A$ and $B$ to the interface, and $w$ is the separation between the perpendiculars.

By Fermat's principle, at the extremum:

$$\frac{\mathrm{d}L}{dx} = 0$$


Computing the derivative:

$$\frac{\mathrm{d}L}{dx} = n_1 \frac{x}{\sqrt{d_1^2 + x^2}} - n_2 \frac{w-x}{\sqrt{d_2^2 + (w-x)^2}}$$


At the extremum, this vanishes. Note that $\sin\theta_1 = x/\sqrt{d_1^2 + x^2}$ and $\sin\theta_2 = (w-x)/\sqrt{d_2^2 + (w-x)^2}$ are the angles of incidence and refraction. Therefore:

$$n_1 \sin\theta_1 = n_2 \sin\theta_2$$


This is **Snell's law**, derived directly from Fermat's principle using calculus of variations.

:::

### Interference and Diffraction

Huygens' principle naturally explains interference and diffraction because it is fundamentally about **superposition of waves**.

#### Young's Double-Slit Experiment

Two slits act as coherent secondary sources (from a common incident wavefront). Each slit emits spherical wavelets. At a point $P$ on a distant screen:

**Path difference**: The distance from slit 1 to $P$ is $r_1$; from slit 2 is $r_2$. The path difference is $\Delta r = r_2 - r_1$.

**Phase difference**: The phase accumulated from each wavelet is $k\Delta r$.

**Interference condition**:

- **Constructive**: $\Delta r = m\lambda$ (phase difference $2\pi m$), $m = 0, \pm 1, \pm 2, \ldots$ → bright fringe.

- **Destructive**: $\Delta r = (m + 1/2)\lambda$ (phase difference $(2m+1)\pi$) → dark fringe.

#### Fresnel Diffraction from an Edge

When light passes an edge, secondary wavelets from all unobstructed points on the wavefront contribute to a point beyond the geometric shadow. The **Fresnel zones** are regions where the phase difference from neighboring zones is $\pi$ (path difference $\lambda/2$). Summing contributions from Fresnel zones shows that:

1. The intensity near the shadow boundary oscillates (Fresnel fringes).
2. Inside the geometric shadow, there is non-zero intensity from contributions outside the shadow.
3. This is purely a **phase interference effect**—the geometric optics picture breaks down.

:::{admonition} Example: Poisson Spot (1826)
:class: example

When Fresnel's diffraction theory predicted that light would appear at the **center of the shadow of a circular disk**, it seemed absurd. The physicist Poisson used this as an argument *against* wave theory. But experimentalist François Arago verified the prediction: there is indeed a bright spot at the center of the disk's shadow. 

This "Poisson spot" was crucial evidence for the wave nature of light and demonstrated the power of Huygens' principle. What seemed like a paradox—light inside the geometric shadow—is actually constructive interference of wavelets from the edge of the obstacle.
:::

### The Eikonal Equation: Connecting Waves to Rays

Maxwell's equations describe electromagnetic waves in a medium with varying refractive index $n(\boldsymbol{r})$. For high-frequency waves (short wavelength relative to variations in $n$), we can use the **geometric optics approximation**.

**Wave equation**:

$$


\nabla^2 \boldsymbol{E} = \frac{n^2}{c^2} \frac{\partial^2 \boldsymbol{E}}{\partial t^2}

$$ (eq-wave-equation)

**WKB ansatz**: Write the field as

$$


\boldsymbol{E}(\boldsymbol{r}, t) = \boldsymbol{E}_0(\boldsymbol{r}) \, \mathrm{e}^{\mathrm{i}(\psi(\boldsymbol{r}) - \omega t)}

$$ (eq-eikonal-ansatz)

where $\psi(\boldsymbol{r})$ is the **eikonal function** (related to optical path length) and $\boldsymbol{E}_0$ is a slowly-varying envelope.

Substituting into Maxwell's equations and keeping only the leading-order (high-frequency) terms, we obtain the **eikonal equation**:

$$


(\nabla\psi)^2 = n(\boldsymbol{r})^2

$$ (eq-eikonal-equation)

**Light rays** are the orthogonal trajectories to surfaces of constant eikonal, pointing in the direction of increasing eikonal.

:::{admonition} Physical Intuition: Eikonal Rays
:class: note

A light ray follows the direction of increasing eikonal:

$$


\frac{\mathrm{d}\boldsymbol{r}}{ds} = \frac{\nabla\psi}{|\nabla\psi|} = \frac{\nabla\psi}{n}

$$


where $s$ is arc length along the ray. Differentiating the eikonal equation and using the above, we recover **Newton's law for light rays** in an inhomogeneous medium:

$$


\frac{\mathrm{d}}{\mathrm{ds}}\left(n \frac{\mathrm{d}\boldsymbol{r}}{ds}\right) = \nabla n

$$ (eq-newtons-law-light)

This shows that light is **bent by the gradient of the refractive index**, just as particles are bent by forces.
:::

### From Huygens to Fermat: The Stationary Phase Connection

Huygens' principle appears to sum contributions from all possible paths (secondary wavelets from all points on the wavefront). Fermat's principle selects only the extremal path. How are they related?

**The limit connection**: Consider wavefront propagation from $A$ to $B$ through an inhomogeneous medium. Each point on an intermediate wavefront can be viewed as a secondary source. The amplitude at $B$ is the superposition:

$$


\psi_B \propto \int \exp(\mathrm{i} k L(A \to \text{intermediate} \to B)) \, d(\text{intermediate})

$$ (eq-huygens-sum)

where $L$ is the optical path length. This is a sum over all possible paths, weighted by phase $\mathrm{e}^{\mathrm{i}kL}$.

**Stationary phase argument**: When the frequency is high (short wavelength, large $k$), the phase $kL$ oscillates rapidly for paths far from an extremum. Only paths where $\delta L = 0$ (Fermat's principle) have slowly-varying phase and contribute constructively. All other paths cancel through destructive interference.

**Result**: Fermat's principle emerges from Huygens' principle in the high-frequency (short-wavelength) limit. **Geometric optics is the classical limit of wave optics.**

| Optics regime | Principle | Limit |
|--------------|-----------|-------|
| Wave (low frequency) | Huygens' superposition of all wavelets | All paths contribute |
| Transition (intermediate) | Stationary phase approximation | Paths near extremum dominate |
| Ray (high frequency) | Fermat's extremal path | Single classical path |

### Summary

- **Huygens' principle**: Secondary wavelets emanate from all points on a wavefront; their superposition gives the next wavefront.

- **Phase and optical path**: Each wavelet carries phase proportional to optical path length $L = \int n \, \mathrm{d}s$. Huygens-Fresnel integral is a sum over paths with phase weights.

- **Interference and diffraction**: Superposition of wavelets naturally produces interference fringes and diffraction patterns, including the famous Poisson spot.

- **Eikonal equation**: Governs ray trajectories in inhomogeneous media; reduces to Newton's law for light rays.

- **Stationary phase limit**: In the high-frequency limit, only paths with extremal optical path survive—Fermat's principle emerges from Huygens' principle through constructive/destructive interference.

- **Precursor to path integrals**: The Huygens-Fresnel integral structure—a sum over paths with phase weights—is foundational to understanding Feynman's path integral in quantum mechanics (§3.2).
