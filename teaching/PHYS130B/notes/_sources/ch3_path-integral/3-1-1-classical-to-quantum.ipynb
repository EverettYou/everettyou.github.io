## Lecture Notes

Wave-particle duality is not a paradox but a profound insight: waves and particles are complementary descriptions of the same quantum phenomenon, selected by the type of measurement. Planck's quantum relations $E = \hbar\omega$ and $p = \hbar k$ unify them, connecting classical action to quantum phase.

### Two Thousand Years of Dispute: The Nature of Light

One of the most fundamental questions in physics asks: **What is light—a particle or a wave?** The answer, we now know, is both. But this resolution took two millennia to achieve, and the struggle shaped modern physics in profound ways.

#### The Corpuscular (Particle) Theory

**Ancient Greece (5th century BCE)**: The Pythagorean school proposed that light consists of a steady stream of **particles** emitted by visible objects, carrying energy and traveling in straight lines (rays) at finite speed.

**17th–18th centuries**: Isaac **Newton** (1704) championed the corpuscular theory in his *Optiks*. He explained:

- **Dispersion**: White light is a mixture of corpuscles of different colors, each refracted at a slightly different angle, producing the spectrum.

- **Geometric optics**: Light rays are particle trajectories following **Fermat's principle**—they travel along the path of extremal optical path length.

- **Refraction**: Corpuscles accelerate/decelerate at interfaces, changing direction.

Newton's theory integrated with his classical mechanics to dominate physics for a century. The particle picture was powerful, mathematically elegant, and explained many phenomena.

:::{admonition} Fermat's Principle (1662)
:class: note

**Pierre de Fermat** stated the fundamental principle of geometric optics:

$$

\delta L = 0, \quad L = \int n(\boldsymbol{r}) \, ds

$$ (eq-fermat-principle)

Light travels along the path for which the **optical path length** (product of refractive index and distance) is extremized. This is not a minimum but an **extremum**—stationary with respect to small variations in the path. Geometrically, light rays are particle trajectories satisfying a variational principle, the **principle of stationary action** in disguise.
:::

#### The Wave Theory Emerges

**Christiaan Huygens (1690)**: Huygens proposed that light is a **wave** propagating through a medium called the **ether**. He introduced:
- The concept of the **wavefront**: a surface of constant phase.

- **Huygens' principle**: each point on a wavefront acts as a secondary source, emitting spherical wavelets. The new wavefront is the coherent superposition of all these wavelets.

Using this principle, Huygens derived the laws of **reflection** and **refraction** and explained phenomena like **Newton's rings** (interference patterns) that corpuscles alone could not account for.

**Augustin-Jean Fresnel (1819)**: The wave theory reached its apex with Fresnel's rigorous mathematical theory of **diffraction**. When Fresnel's theory predicted a bright spot at the center of the shadow of a disk—seemingly absurd—the French physicist **François Arago** experimentally confirmed it (**Poisson spot**, 1826). Wave theory won decisively.

**James Clerk Maxwell (1865)**: Maxwell unified electricity and magnetism into a single theory of **electrodynamics**, predicting that light is an **electromagnetic wave** with speed $c = 1/\sqrt{\mu_0\epsilon_0}$. This matched the measured speed of light perfectly. Wave theory seemed complete.

| Era | Winner | Key Evidence |
|-----|--------|-----|
| 18th century | Particles | Geometric optics, dispersion |
| 19th century | Waves | Interference, diffraction, Maxwell's theory |

#### The Third War: Planck and Einstein

Just as wave theory seemed final, new phenomena forced a return to particles.

**Max Planck (1900)**: Planck solved the **blackbody radiation** problem by proposing that the energy of light is **quantized**:

$$

E_n = n\hbar\omega, \quad n = 0, 1, 2, \ldots

$$ (eq-planck-quantization)

where $\hbar = h/(2\pi) \approx 1.055 \times 10^{-34}$ J·s is the **reduced Planck constant** and $\omega = 2\pi f$ is the **angular frequency**. Light of frequency $f$ comes in discrete quanta, each carrying energy $E = \hbar\omega$.

**Albert Einstein (1905)**: Einstein explained the **photoelectric effect**—the ejection of electrons from a metal when illuminated by light—using Planck's energy quantum. Light consists of **photons**: particles of energy $E = \hbar\omega$ and momentum $p = \hbar k = h/\lambda$, where $k = 2\pi/\lambda$ is the **wavenumber**.

**Arthur Compton (1923)**: Compton's **scattering** experiment provided direct evidence. When X-rays scatter off free electrons, the scattered X-ray wavelength increases—the photon transfers energy and momentum to the electron like two colliding particles. The **Compton formula** confirmed that photons carry momentum $p = \hbar k$.

:::{admonition} The Quantum Relations
:class: important

The breakthrough was recognizing that **waves and particles are complementary descriptions of the same phenomenon**. The quantum relations connect them:

$$

E = \hbar\omega, \quad p = \hbar k

$$ (eq-quantum-relations)

A wave with frequency $\omega$ and wavenumber $k$ is equivalent to a particle with energy $E$ and momentum $p$. The **de Broglie relation** $\lambda = h/p = 2\pi/k$ is the inverse: particles have wavelength.

In the **classical limit** where $\hbar \to 0$, the wavelength becomes imperceptibly small, and the particle behaves like a point mass—classical mechanics is the limit where quantum wavelike behavior is negligible.
:::

### Wave-Particle Duality

By the 1920s, the resolution became clear: light and matter exhibit **complementary** aspects:

**Particle aspects**:
- Discrete energy packets (quanta)
- Definite momentum and energy
- Localized collision cross-sections (Compton scattering)
- Photoelectric effect (energy transfer)

**Wave aspects**:
- Interference and diffraction
- Superposition
- Frequency and wavelength
- Propagation in space

The **which aspect dominates depends on the experimental setup**:
- When we measure position or energy, we observe **particle-like behavior** (energy quantization, localization).
- When we observe wavelike properties (interference, diffraction), we see **wave-like behavior**.
- Never can we see both simultaneously (the **uncertainty principle**).

#### De Broglie's Hypothesis (1924)

**Louis de Broglie** proposed that **all matter possesses wave properties**. An electron with momentum $p$ has de Broglie wavelength

$$

\lambda = \frac{h}{p} = \frac{2\pi\hbar}{p}

$$ (eq-de-broglie)

This was confirmed by **electron diffraction experiments** (Davisson and Germer, 1927): electrons passing through a crystal exhibit interference patterns, just like X-rays. Particles and waves are unified.

:::{admonition} Numerical example: electron wavelength
:class: example

An electron moving at 1% of the speed of light has momentum

$$

p = m_e v \approx (9.11 \times 10^{-31} \text{ kg})(0.01 \times 3 \times 10^8 \text{ m/s}) \approx 2.73 \times 10^{-24} \text{ kg·m/s}

$$

The de Broglie wavelength is

$$

\lambda = \frac{6.626 \times 10^{-34}}{2.73 \times 10^{-24}} \approx 2.4 \times 10^{-10} \text{ m} = 0.24 \text{ nm}

$$

This is comparable to atomic dimensions (Bohr radius $\approx 0.05$ nm), explaining why electron diffraction is observable. A macroscopic baseball (0.1 kg, 30 m/s) has wavelength $\lambda \sim 10^{-34}$ m—unobservably small.
:::

### Action: The Classical-Quantum Bridge

At the heart of both classical and quantum mechanics lies the **action**—a quantity with dimensions of energy times time (angular momentum).

**Classical mechanics**: The principle of **stationary action** (Hamilton's principle) states that a particle follows the path $\boldsymbol{x}(t)$ that makes the action stationary:

$$

S[\boldsymbol{x}] = \int_{t_i}^{t_f} L(\boldsymbol{x}, \dot{\boldsymbol{x}}, t) \, dt

$$ (eq-classical-action)

where $L$ is the **Lagrangian**. The condition $\delta S = 0$ yields the **Euler-Lagrange equations**, the equations of motion.

**Quantum mechanics**: The de Broglie relation tells us to associate a phase with the action:

$$

\text{Phase} = \frac{S}{\hbar}

$$ (eq-action-phase)

The **Planck constant $\hbar$ sets the scale** relating classical action to quantum phase. A small change $\Delta S \sim \hbar$ produces a phase change of order unity—this is the threshold for quantum interference.

| Quantity | Classical | Quantum |
|----------|-----------|--------|
| Fundamental variable | Action $S$ | Phase $\Phi = S/\hbar$ |
| Extremum condition | $\delta S = 0$ | Constructive interference (stationary phase) |
| Typical scale | Large ($S \gg \hbar$) | Small ($S \sim \hbar$) |

### The Classical Limit: $\hbar \to 0$

How does quantum mechanics reduce to classical mechanics? The answer is **Planck's constant $\hbar$ goes to zero**.

In quantum mechanics, a particle explores **all possible paths**, weighted by their phase $e^{iS/\hbar}$. When $\hbar$ is very small:

1. Paths with action differing by $\Delta S$ have phases differing by $\Delta S/\hbar \gg 1$ radian—they oscillate rapidly and **cancel through destructive interference**.

2. Only paths with stationary action $\delta S = 0$ (the **classical path**) have nearly constant phase in a neighborhood of width $\sim \sqrt{\hbar}$ in position—these paths **interfere constructively**.

3. As $\hbar \to 0$, the constructive region shrinks to a point, and the particle is **confined to the classical trajectory**.

:::{admonition} Quantum-to-classical correspondence
:class: seealso

The transition from quantum to classical is continuous:

- **Microscopic ($\hbar$ significant)**: Many paths contribute; wavefunction spreads; uncertainty principle is important.

- **Mesoscopic ($S \sim 10 \hbar$)**: Classical path dominates, but quantum fluctuations are visible. Semiclassical approximation is accurate.

- **Macroscopic ($S \gg \hbar$)**: Classical path absolutely dominates. Quantum effects vanish; classical mechanics is excellent.

For an atomic system ($\hbar \sim 1$ in natural units, $S \sim 10$), quantum mechanics is essential. For a baseball ($S \sim 10^{30}\hbar$), classical mechanics is indistinguishable from quantum mechanics.
:::

### Summary

- **Wave-particle duality**: Light and matter exhibit complementary wave and particle properties. No experiment can reveal both simultaneously.

- **Historical resolution**: Planck's quanta, Einstein's photons, Compton scattering, and de Broglie wavelengths unified the pictures through quantum relations $E = \hbar\omega$ and $p = \hbar k$.

- **Planck constant $\hbar$**: The fundamental scale that connects action (classical) to phase (quantum). Sets the threshold for quantum interference.

- **Classical limit**: As $\hbar \to 0$, quantum mechanics reduces to classical mechanics through destructive interference suppressing paths away from the stationary-action (classical) path.

- **Path integral foundation**: The path integral formulation treats all paths equally, with phases given by $S/\hbar$. This naturally encompasses both wave and particle aspects.
