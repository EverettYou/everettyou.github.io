## Lecture Notes

Knowledge of how a quantum state evolves with time is the key to controlling it. The Schrödinger equation governs this evolution, and for spin systems in magnetic fields, it reveals a remarkably rich physics: spins precess at a natural frequency determined by the field strength, and by applying precisely timed pulses, we can rotate quantum states with exquisite precision. This interplay between natural evolution and engineered control is the foundation of quantum engineering, from atomic clocks to quantum computers.

### Hamiltonian of a Spin in a Magnetic Field

A spin-1/2 particle with **magnetic moment** $\boldsymbol{\mu}$ experiences an energy shift in an external magnetic field $\boldsymbol{B}$.

#### Classical Picture

Classically, the energy of a magnetic dipole in a magnetic field is

$$

E = -\boldsymbol{\mu} \cdot \boldsymbol{B}

$$

For a spin-1/2 particle, the magnetic moment is proportional to the spin:

$$

\boldsymbol{\mu} = \gamma \boldsymbol{S}

$$

where $\gamma$ is the **gyromagnetic ratio**. For an electron, $\gamma = -g_e \mu_B / \hbar$, where $g_e \approx 2$ is the **g-factor** and $\mu_B = e\hbar/(2m_e) \approx 9.28 \times 10^{-24}$ J/T is the **Bohr magneton**.

#### Quantum Hamiltonian

The quantum Hamiltonian is

$$

\hat{H} = -\boldsymbol{\mu} \cdot \boldsymbol{B} = -\gamma \hat{\boldsymbol{\sigma}} \cdot \boldsymbol{B}

$$ (eq-spin-hamiltonian)

where we use the **operator form** $\hat{\boldsymbol{\sigma}} = (\hat{\sigma}^x, \hat{\sigma}^y, \hat{\sigma}^z)$ for the Pauli matrices.

For a magnetic field along the z-axis, $\boldsymbol{B} = B_0 \hat{z}$:

$$

\hat{H} = -\gamma B_0 \hat{\sigma}^z

$$

:::{admonition} Evaluation: Hamiltonian Matrix Form
:class: dropdown note

$$

\hat{H} = -\gamma B_0 \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} = \begin{pmatrix} -\gamma B_0 & 0 \\ 0 & \gamma B_0 \end{pmatrix}

$$ (eq-hamiltonian-z-field)

Eigenvalues: $E_\uparrow = -\gamma B_0$ (spin-up) and $E_\downarrow = +\gamma B_0$ (spin-down).

Energy difference: $\Delta E = E_\downarrow - E_\uparrow = 2\gamma B_0$
:::

:::{admonition} Interpretation
:class: note

A spin aligned with the field (spin-up along z, in the direction of $\boldsymbol{B}$) has lower energy. A spin anti-aligned (spin-down, opposite to $\boldsymbol{B}$) has higher energy. The energy difference increases linearly with magnetic field strength.
:::

### The Larmor Frequency

Define the **Larmor frequency** as

$$

\omega_0 = \gamma B_0

$$ (eq-larmor-frequency)

Then the Hamiltonian becomes

$$

\hat{H} = \frac{\hbar \omega_0}{2} \hat{\sigma}^z

$$ (eq-hamiltonian-larmor)

(absorbing a factor of $\hbar/2$ by convention, as is standard for qubits).

The **Larmor frequency** is the **natural precession frequency** of the spin around the magnetic field direction. It is fundamental in NMR, ESR, and quantum computing.

:::{
admonition} Example: Magnetic Field Strength
:class: example


For a typical NMR field of $B_0 = 1$ Tesla and a proton ($\gamma \approx 2.67 \times 10^8$ rad/(s·T)):

$$

\omega_0 = 2.67 \times 10^8 \times 1 = 2.67 \times 10^8 \text{ rad/s} \approx 43 \text{ MHz}

$$

This is why NMR operates at radiofrequency (RF) wavelengths.
:::

### Time Evolution in a Static Field

For a Hamiltonian $\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z$, the time-evolution operator is

$$

\hat{U}(t) = e^{-i\hat{H}t/\hbar} = e^{-i\omega_0 \hat{\sigma}^z t/2}

$$ (eq-time-evolution-static)

Since $\hat{\sigma}^z$ is diagonal with eigenvalues $\pm 1$, we can evaluate this easily:

$$

\hat{U}(t) = e^{-i\omega_0 t/2} |0\rangle\langle 0| + e^{+i\omega_0 t/2} |1\rangle\langle 1| = \begin{pmatrix} e^{-i\omega_0 t/2} & 0 \\ 0 & e^{+i\omega_0 t/2} \end{pmatrix}

$$ (eq-unitary-static)

#### Initial State: Spin-Up

If we start with $|\psi(0)\rangle = |0\rangle$ (spin-up), then:

$$

|\psi(t)\rangle = \hat{U}(t)|0\rangle = e^{-i\omega_0 t/2}|0\rangle

$$

The state acquires only a **global phase** and doesn't evolve otherwise. It remains spin-up forever (with unit probability).

#### Initial State: Equal Superposition

If we start with $|\psi(0)\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$, then:

$$

|\psi(t)\rangle = \frac{1}{\sqrt{2}}\left(e^{-i\omega_0 t/2}|0\rangle + e^{+i\omega_0 t/2}|1\rangle\right)

$$

The **relative phase** between the two components is $e^{-i\omega_0 t}$. The state precesses around the z-axis at frequency $\omega_0$ on the Bloch sphere.

For the equal superposition state, the expectation values are:

:::{admonition} Calculation: Larmor Precession Expectation Values
:class: dropdown note

Starting from $|\psi(t)\rangle = \frac{1}{\sqrt{2}}(e^{-i\omega_0 t/2}|0\rangle + e^{+i\omega_0 t/2}|1\rangle)$:

$$

\langle \hat{\sigma}^x \rangle(t) = \frac{1}{2}\langle 0|e^{i\omega_0 t/2} + \langle 1|e^{-i\omega_0 t/2} \cdot \hat{\sigma}^x \cdot (e^{-i\omega_0 t/2}|0\rangle + e^{i\omega_0 t/2}|1\rangle)

$$

After simplification:

$$

\langle \hat{\sigma}^x \rangle(t) = \cos(\omega_0 t)

$$

Similarly, $\langle \hat{\sigma}^y \rangle(t) = \sin(\omega_0 t)$ and $\langle \hat{\sigma}^z \rangle(t) = 0$ (unchanged).
:::

The **Bloch vector** rotates around the z-axis at angular velocity $\omega_0$. This is **Larmor precession**.

### Driven Evolution: Rabi Oscillations

In practice, we don't just let the spin precess in a static field. We apply a **driving field** to manipulate the state. This is the basis of quantum control.

#### The Rabi Hamiltonian (Rotating Wave Approximation)

Consider a qubit subject to a **resonant microwave drive** at frequency $\omega_0$ (matching the Larmor frequency). In the **rotating frame** and under the **rotating wave approximation (RWA)**, the Hamiltonian simplifies to:

$$

\hat{H}_{\text{RWA}} = \frac{\hbar\Omega}{2}\hat{\sigma}^x

$$ (eq-rabi-hamiltonian)

where $\Omega$ is the **Rabi frequency**, proportional to the drive field amplitude.

The **rotating frame** eliminates the fast oscillations at $\omega_0$, leaving only the effect of the drive on the qubit.

#### Time Evolution in the RWA

The time-evolution operator is

$$

\hat{U}(t) = e^{-i\Omega\sigma^x t/2}

$$

:::{admonition} Evaluation: Rabi Rotation
:class: dropdown note

$$

\hat{U}(t) = \cos(\Omega t/2) I - i\sin(\Omega t/2)\hat{\sigma}^x

$$ (eq-rabi-unitary)

:::

This is a **rotation around the x-axis** on the Bloch sphere by angle $\Omega t$.

#### Rabi Oscillations: Explicit Example

**Result**: Starting from $|\psi(0)\rangle = |0\rangle$, the system oscillates between $|0\rangle$ and $|1\rangle$ with probability:

$$

P_1(t) = \sin^2(\Omega t/2)

$$

This shows **complete population inversion** at $t = \pi/\Omega$, and returns to the initial state at $t = 2\pi/\Omega$.

:::{admonition} Derivation: Rabi Oscillations
:class: dropdown note

$$

|\psi(t)\rangle = \hat{U}(t)|0\rangle = \left(\cos(\Omega t/2) I - i\sin(\Omega t/2)\hat{\sigma}^x\right)|0\rangle

$$

Since $\hat{\sigma}^x |0\rangle = |1\rangle$:

$$

|\psi(t)\rangle = \cos(\Omega t/2)|0\rangle - i\sin(\Omega t/2)|1\rangle

$$ (eq-rabi-state)

**Probability in state $|1\rangle$**:

$$

P_1(t) = |\langle 1 | \psi(t) \rangle|^2 = \sin^2(\Omega t/2)

$$ (eq-rabi-probability)

:::

**Interpretation**: 
- At $t = 0$: $P_1 = 0$ (definitely in $|0\rangle$)
- At $t = \pi/(2\Omega)$: $P_1 = 1$ (definitely in $|1\rangle$) — a **$\pi/2$-pulse** (or "X-90 gate")
- At $t = \pi/\Omega$: $P_1 = 0$ (back to $|0\rangle$) — a **$\pi$-pulse** (or "X gate", equivalent to $\hat{\sigma}^x$)
- At $t = 2\pi/\Omega$: $P_1 = 0$ (returns to $|0\rangle$) — **one full oscillation period**

The probability oscillates sinusoidally at frequency $\Omega$ (the Rabi frequency).

:::{admonition} Example: Pulse Durations for Quantum Gates
:class: dropdown note

**X-90 gate** ($t = \pi/(2\Omega)$): Creates equal superposition

$$

|\psi\rangle = \cos(\pi/4)|0\rangle - i\sin(\pi/4)|1\rangle = \frac{1}{\sqrt{2}}(|0\rangle - i|1\rangle)

$$

This is a **90° rotation around x-axis**.

**X gate** ($t = \pi/\Omega$): Flips the state completely

$$

|\psi\rangle = \cos(\pi/2)|0\rangle - i\sin(\pi/2)|1\rangle = -i|1\rangle \approx |1\rangle

$$

(global phase is ignored).
:::

### Generalization: Arbitrary Drive Directions

If the drive is applied along the x-axis with Rabi frequency $\Omega$, we can rotate around any direction $\hat{n} = (n_x, n_y, n_z)$ by applying the appropriate combination of drive fields:

$$

\hat{U}(\theta, \hat{n}) = e^{-i\theta \hat{n} \cdot \hat{\boldsymbol{\sigma}}/2} = \cos(\theta/2) I - i\sin(\theta/2) \hat{n} \cdot \hat{\boldsymbol{\sigma}}

$$ (eq-arbitrary-rotation)

This is a **rotation by angle $\theta$ around the axis $\hat{n}$** on the Bloch sphere. Any point on the Bloch sphere can be reached from any other by a suitable unitary rotation.

### Damping and Decoherence (Real Systems)

In real devices, spins don't oscillate forever. Two timescales characterize decoherence:

1. **$T_1$ (spin-lattice relaxation)**: Energy decays to the environment. Excited state $|1\rangle$ decays toward ground state $|0\rangle$ on timescale $T_1$.

2. **$T_2$ (spin-spin dephasing)**: Coherence decays. The relative phase between $|0\rangle$ and $|1\rangle$ is lost on timescale $T_2 \leq 2T_1$.

For Rabi oscillations to be observable, the pulse time must satisfy $\Omega T_2 \gg 1$ (many oscillations within the coherence time).

### Summary

- **Hamiltonian for spin in a field**: $\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z$, where $\omega_0 = \gamma B_0$ is the **Larmor frequency**.
- **Larmor precession**: The spin's Bloch vector precesses around the field direction at the Larmor frequency.
- **Rabi oscillations**: Under resonant driving (RWA), the state oscillates between $|0\rangle$ and $|1\rangle$ at the **Rabi frequency** $\Omega$.
- **Key pulse durations**:
  - $\pi/2$-pulse ($t = \pi/(2\Omega)$): Creates equal superposition
  - $\pi$-pulse ($t = \pi/\Omega$): Flips the state (bit-flip gate)
- **Arbitrary rotations**: Combining drives along different directions enables arbitrary unitary rotations of the Bloch sphere.
- **Decoherence limits**: Real systems decay on timescales $T_1$ and $T_2$, setting ultimate limits on quantum gate fidelity.
