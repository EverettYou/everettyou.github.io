# 1.1 States and Observables

### Overview

The quantum world begins with the simplest possible system: a single two-state particle. In this section, we develop the mathematical language of quantum states, observables, and the Bloch sphere representation that makes the abstract concrete. These ideas form the foundation for everything that follows in quantum mechanics.
