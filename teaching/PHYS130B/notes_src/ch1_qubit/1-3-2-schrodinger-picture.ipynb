## Lecture Notes

### Overview

In quantum mechanics, we must decide: does the state evolve in time, or the observables? The **Schrödinger picture** answers: the state evolves. This is the most intuitive framework for understanding how to control quantum systems—by shaping the state via carefully engineered Hamiltonians.

The heart of the Schrödinger picture is the **time-evolution operator** $\hat{U}(t)$, which encodes how any initial state develops. For spin systems, this operator describes both natural precession in a magnetic field and coherent control via microwave drives. Understanding these two regimes—static precession and resonant driving—is the key to quantum engineering.

### The Time-Evolution Operator

:::{admonition} Schrödinger Equation of Motion
:class: important

The **Schrödinger equation** governs state evolution:

$$
\boxed{\mathrm{i}\hbar \frac{\partial\vert\psi(t)\rangle}{\partial t} = \hat{H}\vert\psi(t)\rangle}
$$ (eq-schrodinger-equation)

:::

For a **time-independent Hamiltonian** $\hat{H}$, this has a formal solution:

$$
\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}
$$ (eq-evolution-operator)

and the state evolves as

$$
\vert\psi(t)\rangle = \hat{U}(t)\vert\psi(0)\rangle
$$ (eq-time-evolved-state)


:::{admonition} Properties of the Time-Evolution Operator
:class: important

1. **Composition**: $\hat{U}(t_1)\hat{U}(t_2) = \hat{U}(t_1 + t_2)$. Multiple evolution steps concatenate.

2. **Unitarity**: $\hat{U}^\dagger(t)\hat{U}(t) = \hat{I}$. The operator preserves inner products and probabilities (probability conservation).

3. **Initial condition**: $\hat{U}(0) = \hat{I}$. At $t=0$, no evolution has occurred.

:::

:::{admonition} Derivation: Composition Property
:class: dropdown information

For a time-independent Hamiltonian, the composition property $\hat{U}(t_1)\hat{U}(t_2) = \hat{U}(t_1 + t_2)$ follows directly from the exponential form:

$$

\hat{U}(t_1)\hat{U}(t_2) = \mathrm{e}^{-\mathrm{i}\hat{H}t_1/\hbar}\,\mathrm{e}^{-\mathrm{i}\hat{H}t_2/\hbar} = \mathrm{e}^{-\mathrm{i}\hat{H}(t_1+t_2)/\hbar} = \hat{U}(t_1+t_2)

$$

where the second equality uses $\mathrm{e}^{\hat{A}}\mathrm{e}^{\hat{B}} = \mathrm{e}^{\hat{A}+\hat{B}}$ when $[\hat{A},\hat{B}]=0$ (here $\hat{A} = -\mathrm{i}\hat{H}t_1/\hbar$ and $\hat{B} = -\mathrm{i}\hat{H}t_2/\hbar$ commute since both are proportional to $\hat{H}$).

This means the time-evolution operators form a **one-parameter group**: the set $\{\hat{U}(t) : t \in \mathbb{R}\}$ is closed under multiplication and every element has an inverse $\hat{U}(t)^{-1} = \hat{U}(-t)$.
:::

:::{admonition} Derivation: Unitarity
:class: dropdown information

Since $\hat{H}$ is Hermitian ($\hat{H}^\dagger = \hat{H}$):

$$

\hat{U}^\dagger(t) = (\mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar})^\dagger = \mathrm{e}^{+\mathrm{i}\hat{H}t/\hbar}

$$

Therefore:

$$

\hat{U}^\dagger(t)\hat{U}(t) = \mathrm{e}^{+\mathrm{i}\hat{H}t/\hbar} \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar} = \hat{I}

$$

:::

### Example: Spin Precession in a Static Magnetic Field

Consider a spin-1/2 particle in a **static magnetic field** $\boldsymbol{B} = B_0\hat{z}$. The Hamiltonian is

$$

\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z

$$ (eq-hamiltonian-static)

where $\omega_0 = \gamma B_0$ is the **Larmor frequency** (the natural precession frequency).

Since $\hat{\sigma}^z$ is diagonal with eigenvalues $\pm 1$, the time-evolution operator is:

$$

\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\omega_0 t\hat{\sigma}^z/2} = \begin{pmatrix} \mathrm{e}^{-\mathrm{i}\omega_0 t/2} & 0 \\ 0 & \mathrm{e}^{+\mathrm{i}\omega_0 t/2} \end{pmatrix}

$$ (eq-unitary-static)

**Starting from spin-up $\vert\uparrow\rangle$.**

$$

\vert\psi(t)\rangle = \hat{U}(t)\vert\uparrow\rangle = \mathrm{e}^{-\mathrm{i}\omega_0 t/2}\vert\uparrow\rangle

$$

The state **remains spin-up** but acquires a **global phase** $\mathrm{e}^{-\mathrm{i}\omega_0 t/2}$. Global phases are unobservable, so the measurement outcomes never change.

**Starting from an equal superposition.**

$$

\vert\psi(0)\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\rangle + \vert\downarrow\rangle) = \vert+\rangle

$$

After time evolution:

$$

\vert\psi(t)\rangle = \frac{1}{\sqrt{2}}(\mathrm{e}^{-\mathrm{i}\omega_0 t/2}\vert\uparrow\rangle + \mathrm{e}^{+\mathrm{i}\omega_0 t/2}\vert\downarrow\rangle)

$$

The **relative phase** between the two components is $\mathrm{e}^{-\mathrm{i}\omega_0 t}$. This causes the Bloch vector to rotate around the z-axis.

:::{admonition} Bloch Vector Motion
:class: dropdown note

The expectation values are:

$$

\langle \hat{\sigma}^x \rangle(t) = \cos(\omega_0 t)

$$

$$

\langle \hat{\sigma}^y \rangle(t) = \sin(\omega_0 t)

$$

$$

\langle \hat{\sigma}^z \rangle(t) = 0

$$

The Bloch vector precesses around the z-axis at the **Larmor frequency** $\omega_0$.

:::

:::{admonition} Derivation: Bloch Vector Expectation Values
:class: dropdown information

We compute $\langle\hat{\sigma}^a\rangle(t)$ for the precessing state $\vert\psi(t)\rangle = \frac{1}{\sqrt{2}}(\mathrm{e}^{-\mathrm{i}\omega_0 t/2}\vert\uparrow\rangle + \mathrm{e}^{+\mathrm{i}\omega_0 t/2}\vert\downarrow\rangle)$.

**$\langle\hat{\sigma}^x\rangle$:** Using $\hat{\sigma}^x = \vert\uparrow\rangle\langle\downarrow\vert + \vert\downarrow\rangle\langle\uparrow\vert$:

$$

\langle\hat{\sigma}^x\rangle = \tfrac{1}{2}(\mathrm{e}^{+\mathrm{i}\omega_0 t/2}\langle\uparrow\vert + \mathrm{e}^{-\mathrm{i}\omega_0 t/2}\langle\downarrow\vert)(\vert\uparrow\rangle\langle\downarrow\vert + \vert\downarrow\rangle\langle\uparrow\vert)(\mathrm{e}^{-\mathrm{i}\omega_0 t/2}\vert\uparrow\rangle + \mathrm{e}^{+\mathrm{i}\omega_0 t/2}\vert\downarrow\rangle)

$$

The cross terms give $\tfrac{1}{2}(\mathrm{e}^{+\mathrm{i}\omega_0 t} + \mathrm{e}^{-\mathrm{i}\omega_0 t}) = \cos(\omega_0 t)$.

**$\langle\hat{\sigma}^y\rangle$:** Using $\hat{\sigma}^y = -\mathrm{i}\vert\uparrow\rangle\langle\downarrow\vert + \mathrm{i}\vert\downarrow\rangle\langle\uparrow\vert$:

$$

\langle\hat{\sigma}^y\rangle = \tfrac{1}{2}(-\mathrm{i}\,\mathrm{e}^{+\mathrm{i}\omega_0 t} + \mathrm{i}\,\mathrm{e}^{-\mathrm{i}\omega_0 t}) = \sin(\omega_0 t)

$$

**$\langle\hat{\sigma}^z\rangle$:** Using $\hat{\sigma}^z = \vert\uparrow\rangle\langle\uparrow\vert - \vert\downarrow\rangle\langle\downarrow\vert$:

$$

\langle\hat{\sigma}^z\rangle = \tfrac{1}{2}(1 - 1) = 0

$$

The Bloch vector traces a circle $\boldsymbol{r}(t) = (\cos\omega_0 t,\, \sin\omega_0 t,\, 0)$ in the $xy$-plane, precessing at frequency $\omega_0$.
:::

### Example: Driven Evolution — Rabi Oscillations

Now suppose we apply a **microwave drive** resonant with the Larmor frequency. In the **rotating frame** and under the **rotating-wave approximation**, the Hamiltonian becomes

$$

\hat{H}_{\text{RWA}} = \frac{\hbar\Omega}{2}\hat{X}

$$ (eq-rabi-hamiltonian)

where $\Omega$ is the **Rabi frequency** (proportional to the drive field amplitude). Here we switch to QI notation ($\hat{X}$, $\hat{Z}$ with $\vert 0\rangle$, $\vert 1\rangle$) since Rabi oscillations are the basis of quantum gate operations.


:::{admonition} Derivation: The Rotating-Wave Approximation
:class: dropdown information

We derive the RWA Hamiltonian $\hat{H}_{\mathrm{RWA}} = \frac{\hbar\Omega}{2}\hat{X}$ from the full lab-frame Hamiltonian for a spin-1/2 in a static field $B_0\hat{z}$ plus a linearly-polarized oscillating drive:

$$

\hat{H}_{\mathrm{lab}} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z + \hbar\Omega\cos(\omega t)\hat{\sigma}^x

$$

where $\omega_0$ is the Larmor frequency, $\omega$ is the drive frequency, and $\Omega$ is the drive amplitude (Rabi frequency).

**Step 1: Rotating frame.** Define $\hat{U}_R(t) = \mathrm{e}^{-\mathrm{i}\omega t\hat{\sigma}^z/2}$. The Schrödinger equation in the new frame is governed by the effective Hamiltonian:

$$

\hat{H}_R = \hat{U}_R^\dagger \hat{H}_{\mathrm{lab}} \hat{U}_R - \mathrm{i}\hbar \hat{U}_R^\dagger \dot{\hat{U}}_R

$$

The **frame-correction term** is $-\mathrm{i}\hbar\hat{U}_R^\dagger\dot{\hat{U}}_R = -\frac{\hbar\omega}{2}\hat{\sigma}^z$ (since $\hat{U}_R$ commutes with $\hat{\sigma}^z$). Combined with the static-field term:

$$

\hat{U}_R^\dagger\frac{\hbar\omega_0}{2}\hat{\sigma}^z\hat{U}_R - \frac{\hbar\omega}{2}\hat{\sigma}^z = \frac{\hbar\Delta}{2}\hat{\sigma}^z

$$

where $\Delta = \omega_0 - \omega$ is the **detuning**.

**Step 2: Rotate the drive term.** The Pauli rotation identity (the Euler formula for half-angle) gives:

$$

\mathrm{e}^{+\mathrm{i}\omega t\hat{\sigma}^z/2}\,\hat{\sigma}^x\,\mathrm{e}^{-\mathrm{i}\omega t\hat{\sigma}^z/2} = \cos(\omega t)\hat{\sigma}^x + \sin(\omega t)\hat{\sigma}^y

$$

Substituting into the drive term and using double-angle identities:

$$

\hat{U}_R^\dagger[\hbar\Omega\cos(\omega t)\hat{\sigma}^x]\hat{U}_R = \frac{\hbar\Omega}{2}\hat{\sigma}^x + \underbrace{\frac{\hbar\Omega}{2}[\cos(2\omega t)\hat{\sigma}^x + \sin(2\omega t)\hat{\sigma}^y]}_{\text{counter-rotating (CR) terms, frequency }2\omega}

$$

**Step 3: Apply the RWA.** The full rotating-frame Hamiltonian is:

$$

\hat{H}_R = \frac{\hbar\Delta}{2}\hat{\sigma}^z + \frac{\hbar\Omega}{2}\hat{\sigma}^x + \hat{H}_{\mathrm{CR}}(t)

$$

The **rotating-wave approximation** drops $\hat{H}_{\mathrm{CR}}$ because its $2\omega$ oscillations average to zero over the slow Rabi timescale $1/\Omega$ when $\Omega \ll \omega$, yielding:

$$

\hat{H}_{\mathrm{RWA}} = \frac{\hbar\Delta}{2}\hat{\sigma}^z + \frac{\hbar\Omega}{2}\hat{\sigma}^x

$$

At resonance $\Delta = 0$, using QI notation $\hat{\sigma}^x = \hat{X}$: $\hat{H}_{\mathrm{RWA}} = \frac{\hbar\Omega}{2}\hat{X}$.

**Validity.** The RWA requires $\Omega \ll \omega_0$. For a superconducting qubit ($\omega_0/(2\pi) \approx 5$ GHz, $\Omega/(2\pi) \approx 50$ MHz), the ratio $\Omega/\omega_0 \approx 0.01$. The leading correction is the **Bloch-Siegert shift** $\delta\omega_{\mathrm{BS}} \sim \Omega^2/(4\omega_0)$, typically a few kHz.
:::

The time-evolution operator is

$$

\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\Omega t\hat{X}/2} = \cos(\Omega t/2)\hat{I} - \mathrm{i}\sin(\Omega t/2)\hat{X}

$$ (eq-rabi-unitary)

This is a **rotation around the x-axis** by angle $\Omega t$ on the Bloch sphere.

:::{admonition} Derivation: Euler Formula for Pauli Exponentials
:class: dropdown information

Any Pauli matrix $\hat{\sigma}^a$ ($a = x, y, z$) satisfies $(\hat{\sigma}^a)^2 = \hat{I}$, and more generally $(\hat{\boldsymbol{n}}\cdot\hat{\boldsymbol{\sigma}})^2 = \hat{I}$ for any unit vector $\boldsymbol{n}$. This causes the Taylor series of the exponential to collapse.

Write $\hat{A} = \hat{\boldsymbol{n}}\cdot\hat{\boldsymbol{\sigma}}$ with $\hat{A}^2 = \hat{I}$. Then:

$$

\mathrm{e}^{\mathrm{i}\hat{A}\theta} = \sum_{k=0}^{\infty} \frac{(\mathrm{i}\theta)^k}{k!} \hat{A}^k

$$

Since $\hat{A}^{2m} = \hat{I}$ and $\hat{A}^{2m+1} = \hat{A}$, the even and odd terms separate:

$$

\mathrm{e}^{\mathrm{i}\hat{A}\theta} = \hat{I}\sum_{m=0}^{\infty} \frac{(\mathrm{i}\theta)^{2m}}{(2m)!} + \hat{A}\sum_{m=0}^{\infty} \frac{(\mathrm{i}\theta)^{2m+1}}{(2m+1)!} = \cos\theta\,\hat{I} + \mathrm{i}\sin\theta\,\hat{A}

$$

Setting $\hat{A} = \hat{X}$ and $\theta = -\Omega t/2$:

$$

\mathrm{e}^{-\mathrm{i}\Omega t\hat{X}/2} = \cos(\Omega t/2)\,\hat{I} - \mathrm{i}\sin(\Omega t/2)\,\hat{X}

$$

More generally, for any unit vector $\boldsymbol{n}$:

$$

\mathrm{e}^{-\mathrm{i}\boldsymbol{n}\cdot\hat{\boldsymbol{\sigma}}\,\theta/2} = \cos(\theta/2)\,\hat{I} - \mathrm{i}\sin(\theta/2)\,\boldsymbol{n}\cdot\hat{\boldsymbol{\sigma}}

$$

This **Euler formula for Pauli exponentials** is the matrix generalization of Euler's formula $\mathrm{e}^{\mathrm{i}\theta} = \cos\theta + \mathrm{i}\sin\theta$.
:::

**Starting from $\vert 0\rangle$.** The evolved state is

$$

\vert\psi(t)\rangle = \hat{U}(t)\vert 0\rangle = \cos(\Omega t/2)\vert 0\rangle - \mathrm{i}\sin(\Omega t/2)\vert 1\rangle

$$

The **probability of measuring the system in state $\vert 1\rangle$** is:

$$

P_1(t) = \sin^2(\Omega t/2)

$$ (eq-rabi-probability)

This oscillates sinusoidally between 0 and 1 at frequency $\Omega$ — hence "Rabi oscillations."

:::{admonition} Standard Pulse Durations
:class: note

- **$\pi/2$-pulse** ($t = \pi/(2\Omega)$): $P_1 = \sin^2(\pi/4) = 1/2$ → creates equal superposition
- **$\pi$-pulse** ($t = \pi/\Omega$): $P_1 = \sin^2(\pi/2) = 1$ → complete population inversion (flips the state)
- **$2\pi$-pulse** ($t = 2\pi/\Omega$): $P_1 = \sin^2(\pi) = 0$ → returns to initial state (one full cycle)

:::

:::{admonition} Physical Interpretation
:class: note

Rabi oscillations are the mechanism of **quantum gates**. By timing the microwave pulse to duration $t = \pi/(2\Omega)$ or $t = \pi/\Omega$, we can create any rotation we want on the Bloch sphere. This is how quantum computers manipulate qubits.

:::


:::{admonition} Discussion
:class: dropdown tip

**What limits quantum control in practice?**

In theory, we can implement any unitary gate on a qubit by choosing the right Hamiltonian and pulse duration. A $\pi$-pulse flips $\vert 0\rangle \to \vert 1\rangle$; a $\pi/2$-pulse creates a perfect superposition.

But real qubits are never perfectly isolated:
- **Decoherence** destroys superpositions on a timescale $T_2$. If the gate time $t_{\text{gate}} \sim 1/\Omega$ is not much shorter than $T_2$, the operation is imperfect.
- **Leakage**: real atoms have more than two levels. A strong drive can excite higher states outside the qubit subspace.
- **Calibration errors**: the Rabi frequency $\Omega$ depends on the drive amplitude, which fluctuates.

For a superconducting transmon qubit with $T_2 \sim 100\,\mu$s and gate time $\sim 20$ ns, roughly how many gates can you perform before decoherence ruins the computation? What does this imply for the feasibility of quantum error correction?
:::

### Summary

- **Schrödinger picture**: The quantum state $\vert\psi(t)\rangle$ evolves; observables are fixed.
- **Time-evolution operator**: $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$ is unitary and preserves probability.
- **Larmor precession**: In a static magnetic field $B_0$, the spin precesses around $\boldsymbol{B}$ at the **Larmor frequency** $\omega_0 = \gamma B_0$. The Bloch vector rotates on the Bloch sphere.
- **Rabi oscillations**: A resonant microwave drive with Rabi frequency $\Omega$ causes the state to oscillate between $\vert 0\rangle$ and $\vert 1\rangle$ as $P_1(t) = \sin^2(\Omega t/2)$.
- **Quantum gates**: By precisely timing pulses, we can perform arbitrary rotations on the Bloch sphere and implement quantum gates.
