# 1.1.1 What is a Qubit

## Lecture Notes

### Overview

Quantum mechanics is famously abstract. Unlike gravity or fluid flow, you cannot build intuition from everyday experience—no one has ever "seen" a superposition or "felt" an entangled state. The quantum world operates on scales and principles so far from daily life that we cannot simply model it by analogy. Instead, we must rely on mathematical structures and learn what nature does from observation data.

This is not as alien as it sounds. Modern machine learning has shown us a powerful paradigm: **everything is a vector.** Images, words, proteins, music—all are embedded as vectors in high-dimensional spaces, and all transformations (classification, generation, translation) are operators acting on those vectors. Quantum mechanics arrived at the same insight a century earlier: every physical state is a vector, and every physical process—measurement, time evolution, symmetry—is an operator.

### What is Physics? Description and Prediction

:::{admonition} Two Pillars of Physics
:class: note

Physics rests on two fundamental objectives:

1. **Description**: How do we represent the state of a physical system?
2. **Prediction**: Given a state, how do we predict the outcome of an experiment?

For a classical system (a ball rolling down a ramp), we describe the state by specifying position and velocity. We predict the future using Newton's laws.

For a quantum system, the description and prediction mechanisms are radically different.
:::

**Quantum mechanics answers:**

| Aspect | Classical | Quantum |
|--------|-----------|---------|
| **Description** | Definite values: $x$, $v$, $E$ | Abstract vector in Hilbert space: $\vert \psi\rangle$ |
| **Prediction** | Equations of motion (Newton, Lagrange) | Operators acting on vectors; time evolution (Schrödinger) |
| **Role of measurement** | Non-invasive: observe without disturbing | Invasive: measurement collapses state and changes outcome |
| **Information** | Deterministic: same initial state → same outcome | Probabilistic: same state → distribution of outcomes |

### The Qubit: Mathematical Definition

:::{admonition} Qubit Definition
:class: important

A **qubit** is a quantum system that lives in a **two-dimensional Hilbert space** over the complex numbers. Any pure state is a unit vector (a "ket"):

$$

|\psi\rangle = \alpha|0\rangle + \beta|1\rangle, \quad |\alpha|^2 + |\beta|^2 = 1

$$ (eq-qubit-state)

where $\vert 0\rangle$ and $\vert 1\rangle$ are orthonormal basis states, and $\alpha, \beta \in \mathbb{C}$.
:::

**Key properties:**

- **Basis states** $\vert 0\rangle$ and $\vert 1\rangle$ are orthonormal: $\langle 0\vert 1\rangle = 0$, $\langle 0\vert 0\rangle = 1$.
- **Superposition**: Unlike a classical bit (which is 0 or 1), a qubit can be in a superposition of both. The coefficients $\alpha$ and $\beta$ describe the quantum state.
- **Normalization**: The condition $|\alpha|^2 + |\beta|^2 = 1$ ensures the state is normalized. We interpret $|\alpha|^2$ as the probability of measuring the qubit in state $\vert 0\rangle$.
- **Global phase**: States that differ by a global phase (e.g., $\vert \psi\rangle$ and $\mathrm{e}^{\mathrm{i}\phi}\vert \psi\rangle$) are physically identical. Thus a qubit is truly a two-parameter system (after factoring out global phase and normalization).

:::{admonition} Derivation: Global Phase is Unobservable
:class: dropdown information

Let $\vert\psi'\rangle = \mathrm{e}^{\mathrm{i}\gamma}\vert\psi\rangle$ for any real $\gamma$. For any measurement outcome $\vert m\rangle$, the Born rule gives:

$$
P(m \vert \psi') = \vert\langle m\vert\psi'\rangle\vert^2 = \vert\langle m\vert\mathrm{e}^{\mathrm{i}\gamma}\psi\rangle\vert^2 = \vert\mathrm{e}^{\mathrm{i}\gamma}\vert^2\,\vert\langle m\vert\psi\rangle\vert^2 = \vert\langle m\vert\psi\rangle\vert^2 = P(m \vert \psi)
$$

Since $\vert\mathrm{e}^{\mathrm{i}\gamma}\vert = 1$ for any real $\gamma$, every measurement probability is unchanged. Because all physical information about a quantum state is encoded in measurement probabilities, no experiment can ever distinguish $\vert\psi\rangle$ from $\mathrm{e}^{\mathrm{i}\gamma}\vert\psi\rangle$.

This freedom lets us fix the global phase arbitrarily — for example, by requiring the coefficient of $\vert 0\rangle$ to be real and non-negative. Doing so removes one real parameter from the state description, reducing the four real numbers $\{\mathrm{Re}\,\alpha,\, \mathrm{Im}\,\alpha,\, \mathrm{Re}\,\beta,\, \mathrm{Im}\,\beta\}$ to three, and then normalization removes one more, leaving exactly two free real parameters.
:::

### Physical Realizations of Qubits

A qubit is not merely a mathematical abstraction. Many physical systems naturally obey the two-state model:

| Platform | $\vert 0\rangle$ | $\vert 1\rangle$ | $\vert 0\rangle$–$\vert 1\rangle$ splitting | Gates via |
|----------|------|------|------|------|
| **Electron spin** | $\vert \downarrow\rangle$ | $\vert \uparrow\rangle$ | $\mu_B B \sim 0.1$ meV (at $\sim$1 T) | RF/microwave pulses |
| **Photon polarization** | horizontal $\vert H\rangle$ | vertical $\vert V\rangle$ | 0 (degenerate) | Waveplates (passive optics) |
| **Trapped ion** | hyperfine ground $\vert g\rangle$ | excited $\vert e\rangle$ | $\sim\mu$eV (GHz hyperfine) | Laser pulses |
| **Superconducting circuit** | transmon ground | first excited | $\sim\mu$eV ($\sim$5 GHz) | Microwave pulses |
| **Nuclear spin** | $m_I = -1/2$ | $m_I = +1/2$ | $\mu_N B \sim$ neV (at $\sim$1 T) | RF pulses (NMR) |

**Why the same mathematics everywhere:** All two-level quantum systems obey the same Schrödinger equation with a $2 \times 2$ Hamiltonian. The specific physics differs (magnetic fields, photonic transitions, circuit dynamics), but the abstract qubit rules are universal.

:::{admonition} Discussion
:class: dropdown tip

**Why do physicists care about two-level systems when nature has many more states?**

A realistic electron has both orbital angular momentum and spin, giving many possible energy levels. A real atom has infinitely many states (different orbitals). Why restrict attention to qubits?

- In many experimental conditions (low temperature, weak fields, far from resonance with other transitions), only the two lowest energy levels are significantly populated. The **two-level approximation** is valid when $k_B T \ll \Delta E$ and the driving field frequency is resonant with only one transition.
- The qubit is the **minimal quantum system** that exhibits superposition, entanglement, measurement, and unitary evolution. All quantum computation can (in principle) be built from qubits.
- Show that if you restrict a three-level system to only two of its three states, you lose nothing: any unitary transformation on the two-level subspace can be applied without mixing in the third state (under what conditions is this true?).
:::

### Classical Bit vs. Quantum Qubit

| Property | Classical Bit | Quantum Qubit |
|----------|---------------|---------------|
| **Values** | 0 or 1 | $\alpha\vert 0\rangle + \beta\vert 1\rangle$ (superposition) |
| **State space** | 2 possible values | Infinitely many (Bloch sphere: $\mathbb{CP}^1$) |
| **Measurement** | Non-invasive; outcome pre-exists | Invasive; outcome is random, state collapses |
| **Repeatability** | Measuring the same bit twice gives the same result | Measuring the same qubit twice gives the same result, but the first measurement destroys the superposition |
| **Copying** | Trivial: copy the bit | **Impossible**: no-cloning theorem forbids copying an unknown quantum state |
| **Information capacity** | 1 bit of classical information | $\leq 1$ bit of classical information per measurement, but encodes continuous parameters |

:::{admonition} Derivation: No-Cloning Theorem
:class: dropdown information

**Claim.** There is no unitary operator $\hat{U}$ on a two-qubit Hilbert space such that

$$
\hat{U}\vert\psi\rangle\vert 0\rangle = \vert\psi\rangle\vert\psi\rangle \quad \text{for all }\vert\psi\rangle.
$$

**Proof (by linearity of quantum mechanics).**

Suppose such a $\hat{U}$ existed. Apply it to the two computational basis states:

$$
\hat{U}\vert 0\rangle\vert 0\rangle = \vert 0\rangle\vert 0\rangle, \qquad \hat{U}\vert 1\rangle\vert 0\rangle = \vert 1\rangle\vert 1\rangle.
$$

Now consider the superposition state $\vert+\rangle = \tfrac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$. By the assumed cloning property:

$$
\hat{U}\vert+\rangle\vert 0\rangle = \vert+\rangle\vert+\rangle = \frac{1}{2}(\vert 0\rangle + \vert 1\rangle)(\vert 0\rangle + \vert 1\rangle) = \frac{1}{2}(\vert 00\rangle + \vert 01\rangle + \vert 10\rangle + \vert 11\rangle).
$$

But quantum mechanics is **linear**: since $\vert+\rangle\vert 0\rangle = \tfrac{1}{\sqrt{2}}(\vert 0\rangle\vert 0\rangle + \vert 1\rangle\vert 0\rangle)$, unitarity demands:

$$
\hat{U}\vert+\rangle\vert 0\rangle = \frac{1}{\sqrt{2}}\bigl(\hat{U}\vert 0\rangle\vert 0\rangle + \hat{U}\vert 1\rangle\vert 0\rangle\bigr) = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle).
$$

These two outputs are **different**:

$$
\frac{1}{2}(\vert 00\rangle + \vert 01\rangle + \vert 10\rangle + \vert 11\rangle) \neq \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle).
$$

The left side is a product state; the right side is a maximally entangled Bell state. This contradicts the assumption that $\hat{U}$ clones all states. $\blacksquare$

**Alternative proof via inner products.** Suppose $\hat{U}$ clones two states $\vert\psi\rangle$ and $\vert\phi\rangle$:

$$
\hat{U}\vert\psi\rangle\vert 0\rangle = \vert\psi\rangle\vert\psi\rangle, \qquad \hat{U}\vert\phi\rangle\vert 0\rangle = \vert\phi\rangle\vert\phi\rangle.
$$

Since $\hat{U}$ is unitary, it preserves inner products:

$$
\langle\phi\vert\psi\rangle = \langle\phi\vert\langle 0\vert \hat{U}^\dagger\hat{U} \vert\psi\rangle\vert 0\rangle = \langle\phi\vert\phi\vert\psi\rangle\vert\psi\rangle = (\langle\phi\vert\psi\rangle)^2.
$$

Setting $c = \langle\phi\vert\psi\rangle$: the equation $c = c^2$ has only two solutions, $c = 0$ or $c = 1$. Therefore a cloning machine can only clone states that are **identical** ($c=1$) or **perfectly distinguishable** ($c=0$). An arbitrary unknown state $\vert\psi\rangle$ cannot be cloned.
:::



:::{admonition} Derivation: Complex Amplitudes Are Necessary — Real Amplitudes Cannot Describe All Qubit States
:class: dropdown information

**Claim.** Describing a qubit requires complex amplitudes $\alpha, \beta \in \mathbb{C}$. Restricting to real amplitudes $\alpha, \beta \in \mathbb{R}$ is physically insufficient.

**Proof.** For a real-amplitude state $\vert\psi\rangle = \alpha\vert 0\rangle + \beta\vert 1\rangle$ with $\alpha, \beta \in \mathbb{R}$, compute the expectation value of $\hat{\sigma}^y$:

$$
\langle\hat{\sigma}^y\rangle = \langle\psi\vert\hat{\sigma}^y\vert\psi\rangle
= \begin{pmatrix}\alpha & \beta\end{pmatrix}\begin{pmatrix}0 & -\mathrm{i}\\\mathrm{i} & 0\end{pmatrix}\begin{pmatrix}\alpha\\\beta\end{pmatrix}
= \begin{pmatrix}\alpha & \beta\end{pmatrix}\begin{pmatrix}-\mathrm{i}\beta\\ \mathrm{i}\alpha\end{pmatrix}
= -\mathrm{i}\alpha\beta + \mathrm{i}\beta\alpha = 0
$$

for any real $\alpha, \beta$. So **every real-amplitude state gives $\langle\hat{\sigma}^y\rangle = 0$**.

By the Born rule, the Y-measurement probability is $P(Y{=}+1) = (1 + \langle\hat{\sigma}^y\rangle)/2 = 1/2$ for all real-amplitude states. But the Y eigenstate $\vert\mathrm{i}\rangle = \tfrac{1}{\sqrt{2}}(\vert 0\rangle + \mathrm{i}\vert 1\rangle)$ requires $P(Y{=}+1) = 1$. Since no real-amplitude state can achieve this, **real amplitudes cannot represent the full set of qubit states**.

**What complex numbers add.** Allowing $\alpha, \beta \in \mathbb{C}$ introduces a **relative phase** $\phi = \arg(\beta) - \arg(\alpha)$. This phase controls the Y-component of the Bloch vector:

$$
\langle\hat{\sigma}^y\rangle = \sin\theta\sin\phi
$$

(from §1.1.2). When $\phi = \pi/2$, we get $\langle\hat{\sigma}^y\rangle = \sin\theta \neq 0$ — only achievable with a complex relative phase. Restricting to $\phi \in \{0, \pi\}$ (real amplitudes) reduces the Bloch sphere to the equatorial great circle in the XZ-plane, losing the entire Y direction.

**Physical implication.** Any experiment that distinguishes $\vert\mathrm{i}\rangle$ from $\vert +\rangle$ (Y-basis measurement) requires quantum mechanics to use $\mathbb{C}^2$, not $\mathbb{R}^2$.
:::

:::{admonition} Derivation: Superposition Is Not a Classical Probability Mixture
:class: dropdown information

A quantum superposition $\vert\psi\rangle = \alpha\vert 0\rangle + \beta\vert 1\rangle$ differs fundamentally from a classical probability distribution over the outcomes $\{0, 1\}$. We make this precise by showing that a quantum superposition exhibits **interference**, which a classical mixture cannot.

**Setup.** Compare two states with equal probabilities for $\vert 0\rangle$ and $\vert 1\rangle$:

1. **Quantum superposition**: $\vert\psi_\phi\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \mathrm{e}^{\mathrm{i}\phi}\vert 1\rangle)$
2. **Classical mixture**: $\hat{\rho}_{\text{mix}} = \frac{1}{2}\vert 0\rangle\langle 0\vert + \frac{1}{2}\vert 1\rangle\langle 1\vert$

Both give the same probabilities $P(0) = P(1) = 1/2$ when measured in the $\{\vert 0\rangle, \vert 1\rangle\}$ basis.

**Test: Measure in the X-basis.**

Let $\vert +\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$ and $\vert -\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle - \vert 1\rangle)$.

**Superposition**: 

$$
P(+\vert\psi_\phi) = \vert\langle +\vert\psi_\phi\rangle\vert^2 = \left\vert\frac{1}{\sqrt{2}}\cdot\frac{1}{\sqrt{2}}(1 + \mathrm{e}^{\mathrm{i}\phi})\right\vert^2 = \frac{\vert 1 + \mathrm{e}^{\mathrm{i}\phi}\vert^2}{4} = \frac{2 + 2\cos\phi}{4} = \cos^2\!\frac{\phi}{2}
$$

This **oscillates** with the phase $\phi$: from $P(+) = 1$ when $\phi = 0$ to $P(+) = 0$ when $\phi = \pi$.

**Classical mixture**:

$$
P(+\vert\hat{\rho}_{\text{mix}}) = \text{Tr}(\vert +\rangle\langle +\vert\,\hat{\rho}_{\text{mix}}) = \frac{1}{2}\vert\langle +\vert 0\rangle\vert^2 + \frac{1}{2}\vert\langle +\vert 1\rangle\vert^2 = \frac{1}{2}\cdot\frac{1}{2} + \frac{1}{2}\cdot\frac{1}{2} = \frac{1}{2}
$$

This is **always** $\frac{1}{2}$, regardless of $\phi$ — no interference.

**Conclusion.** The phase $\phi$ is physically observable (through measurements in a different basis), even though it doesn't affect measurements in the $\{\vert 0\rangle, \vert 1\rangle\}$ basis. This is the defining feature of **quantum coherence**: relative phases in superpositions create interference effects that have no classical analogue. A qubit state is a superposition, not a mixture — $\phi$ is real physics, not bookkeeping.
:::
:::{admonition} Derivation: Measurement Inevitably Disturbs Non-Orthogonal States
:class: dropdown information

**Theorem.** Any device that reliably distinguishes two quantum states $\vert\psi_1\rangle$ and $\vert\psi_2\rangle$ while leaving each undisturbed must have $\langle\psi_1\vert\psi_2\rangle = 0$. Equivalently, non-orthogonal states cannot both be measured without disturbance.

**Proof.** Suppose a device operates as: given input $\vert\psi_i\rangle$ ($i = 1, 2$), it always outputs outcome $i$ and the post-measurement state remains $\vert\psi_i\rangle$.

**Step 1 — Each state must be an eigenstate.** For input $\vert\psi_1\rangle$ to yield outcome $m_1$ with certainty and leave the state unchanged, $\vert\psi_1\rangle$ must be an eigenstate of the measurement observable $\hat{M}$ with eigenvalue $m_1$:

$$
\hat{M}\vert\psi_1\rangle = m_1\vert\psi_1\rangle, \qquad \hat{M}\vert\psi_2\rangle = m_2\vert\psi_2\rangle, \quad m_1 \neq m_2.
$$

(An eigenstate yields a definite outcome and is unchanged by that measurement — these two properties are equivalent.)

**Step 2 — Different eigenvalues force orthogonality.** Since $\hat{M}$ is Hermitian, compute:

$$
m_1\langle\psi_1\vert\psi_2\rangle = \langle\hat{M}\psi_1\vert\psi_2\rangle = \langle\psi_1\vert\hat{M}\psi_2\rangle = m_2\langle\psi_1\vert\psi_2\rangle.
$$

Since $m_1 \neq m_2$, we conclude $\langle\psi_1\vert\psi_2\rangle = 0$. $\blacksquare$

**Contrapositive.** If $\langle\psi_1\vert\psi_2\rangle \neq 0$, no non-disturbing device can reliably distinguish them. Any successful discrimination must disturb at least one state.

**Comparison to no-cloning.** Both theorems trace to the linearity of quantum mechanics applied to non-orthogonal states. No-cloning says we cannot copy an unknown qubit state; measurement disturbance says we cannot read it out without changing it. The two results reinforce each other: quantum information is neither copyable nor extractable without back-action.

**Classical contrast.** A classical bit can always be read non-invasively — $0$ and $1$ are perfectly distinguishable and orthogonal. For qubits, only mutually orthogonal states can be distinguished and left undisturbed. Any other pair, such as $\vert 0\rangle$ and $\vert+\rangle = (\vert 0\rangle+\vert 1\rangle)/\sqrt{2}$ (with overlap $1/\sqrt{2}$), will be disturbed by any reliable measurement that distinguishes them.
:::

### Summary

- **Physics is about description and prediction**: Quantum mechanics describes systems as vectors in Hilbert space and predicts outcomes using operators and the Born rule.
- **A qubit is the simplest quantum system**: It lives in a two-dimensional complex vector space and can exist in a superposition of basis states $\vert 0\rangle$ and $\vert 1\rangle$.
- **Qubits are physically universal**: Electron spins, photon polarizations, trapped atoms, superconducting circuits, and many other platforms all function as qubits—the mathematics is identical even though the physics varies.
- **Measurement is invasive in quantum mechanics**: Unlike classical measurement, measuring a qubit collapses its superposition and changes the state for future measurements.
- **Qubits form the foundation of quantum information**: All quantum computation and quantum information processing is built from qubits and quantum gates operating on them.
