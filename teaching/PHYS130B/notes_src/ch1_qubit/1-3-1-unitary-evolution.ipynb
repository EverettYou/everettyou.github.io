## Lecture Notes

Quantum systems do not stand still. They evolve in time according to the Schrödinger equation, a fundamental postulate of quantum mechanics. This evolution is governed by the Hamiltonian operator, which encodes the energy and symmetries of the system. A remarkable feature of quantum evolution is that it must be **unitary**: it preserves the normalization of quantum states and the probabilities of all outcomes. This section develops the theory of unitary operators, derives the time-dependent Schrödinger equation, and solves it to obtain the time-evolution operator.

### Unitary Operators

In quantum mechanics, time evolution must satisfy a stringent requirement: it must preserve the normalization of states. If $|\psi(0)\rangle$ is normalized, then $|\psi(t)\rangle$ must remain normalized for all times $t$. This is achieved by **unitary operators**.

#### Definition and Properties

:::{admonition} Unitary Operators
:class: important

A linear operator $\hat{U}$ is **unitary** if

$$


\hat{U}^\dagger \hat{U} = \hat{U}\hat{U}^\dagger = I

$$ (eq-unitarity)

where $\hat{U}^\dagger$ is the **adjoint** (conjugate transpose) and $I$ is the identity.
:::

**Key properties of unitary operators:**

1. **Inverse**: $\hat{U}^{-1} = \hat{U}^\dagger$. Unitary operators are always invertible.

2. **Preserve inner products**: If $\hat{U}$ is unitary, then

$$


\langle \psi | \hat{U}^\dagger \hat{U} | \phi \rangle = \langle \psi | \phi \rangle

$$


This means $\hat{U}$ preserves the overlap between any two states.

3. **Preserve normalization**: For a normalized state $|\psi\rangle$ with $\langle \psi | \psi \rangle = 1$,

$$


\langle \psi | \hat{U}^\dagger \hat{U} | \psi \rangle = \langle \psi | \psi \rangle = 1

$$


So $|\psi'\rangle = \hat{U}|\psi\rangle$ is also normalized.

:::{admonition} Discussion: Unitarity Constraint
:class: dropdown warning

The time-evolution operator $\hat{U}(t)$ must be unitary to preserve probability. But unitarity alone is not the only constraint on time evolution. Consider these additional requirements:

- **Linearity**: Suppose $|\psi_1\rangle \to \hat{U}|\psi_1\rangle$ and $|\psi_2\rangle \to \hat{U}|\psi_2\rangle$. Why must $\alpha|\psi_1\rangle + \beta|\psi_2\rangle \to \alpha\hat{U}|\psi_1\rangle + \beta\hat{U}|\psi_2\rangle$? Is this a consequence of unitarity, or an additional postulate? (Hint: think about superposition being fundamental in quantum mechanics.)

- **Composition law**: If evolution from $t=0$ to $t_1$ is given by $\hat{U}(t_1)$, and from $t_1$ to $t_2$ by $\hat{U}(t_2-t_1)$, the total must be $\hat{U}(t_2) = \hat{U}(t_2-t_1)\hat{U}(t_1)$. Show that if we require $\hat{U}(t)$ to satisfy this composition law, be unitary, and satisfy $\hat{U}(0) = \boldsymbol{I}$, then $\hat{U}(t)$ must have the form $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{G}t}$ for some operator $\hat{G}$. (Hint: differentiate $\hat{U}(t)\hat{U}(s) = \hat{U}(t+s)$ with respect to $t$ at $t=0$.)

- **Hermitian generator**: If $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{G}t}$ is unitary for all real $t$, what must be true about $\hat{G}$? Differentiate $\hat{U}^\dagger(t)\hat{U}(t) = \boldsymbol{I}$ and show that $\hat{G}$ must be Hermitian.
:::

4. **Preserve probabilities**: Measurement probabilities are unchanged. If $\hat{O}$ is an observable with eigenvalues $\lambda_i$ and eigenvectors $|\phi_i\rangle$, then measuring $\hat{O}$ on the evolved state $\hat{U}|\psi\rangle$ gives the same probability distribution as measuring $\hat{U}^\dagger \hat{O} \hat{U}$ on the original state.

:::{admonition} Unitarity: Preserves Probability
:class: dropdown note

Consider the probability of outcome $\lambda_i$ when measuring $\hat{O}$ on state $\hat{U}|\psi\rangle$:

$$


P_i(t) = |\langle \phi_i | \hat{U} | \psi \rangle|^2 = \langle \psi | \hat{U}^\dagger | \phi_i \rangle \langle \phi_i | \hat{U} | \psi \rangle

$$


By linearity and unitarity ($\hat{U}^\dagger \hat{U} = I$), this equals

$$


\langle \psi | \hat{U}^\dagger \hat{O}_i \hat{U} | \psi \rangle

$$


where $\hat{O}_i = |\phi_i\rangle\langle \phi_i|$ is the projector onto the $i$-th eigenstate. The probability distribution depends only on the evolved state through the evolved observable $\hat{U}^\dagger \hat{O} \hat{U}$, preserving all measurement statistics.
:::

### Hermitian Generators and Unitary Operators

A profound connection exists between **Hermitian operators** and **unitary operators**. Any unitary operator can be written as the exponential of a Hermitian operator multiplied by an imaginary constant.

#### The Exponential of an Operator

For a square matrix $\hat{A}$, the exponential $\mathrm{e}^{\hat{A}}$ is defined via the Taylor series:

$$


\mathrm{e}^{\hat{A}} = I + \hat{A} + \frac{\hat{A}^2}{2!} + \frac{\hat{A}^3}{3!} + \cdots = \sum_{n=0}^\infty \frac{\hat{A}^n}{n!}

$$ (eq-operator-exponential)

#### Unitary from Hermitian Generator

:::{admonition} Unitary: Hermitian Generator
:class: important

If $\hat{G}$ is Hermitian ($\hat{G}^\dagger = \hat{G}$) and $\theta$ is a real number, then

$$


\hat{U}(\theta) = \mathrm{e}^{\mathrm{i}\hat{G}\theta}

$$ (eq-unitary-from-hermitian)

is unitary.
:::

:::{admonition} Verification
:class: dropdown note

To verify, compute:

$$


\hat{U}^\dagger \hat{U} = \mathrm{e}^{-\mathrm{i}\hat{G}\theta} \mathrm{e}^{\mathrm{i}\hat{G}\theta} = \mathrm{e}^0 = I

$$


The factor $i$ in the exponent ensures unitarity: the exponent $\mathrm{i}\hat{G}\theta$ is **anti-Hermitian** (its adjoint is the negative of itself), and exponentiating an anti-Hermitian operator yields a unitary matrix.
:::

:::{admonition} Example: Pauli Rotations
:class: example

The Pauli operator $\hat{\sigma}^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ is Hermitian. The unitary operator

$$


\hat{U}(\theta) = \mathrm{e}^{-\mathrm{i}\hat{\sigma}^z \theta / 2}

$$


represents a rotation about the z-axis by angle $\theta$ in the Bloch sphere. Expanding via the exponential:

$$


\mathrm{e}^{-\mathrm{i}\hat{\sigma}^z \theta / 2} = \cos(\theta/2) I - i\sin(\theta/2) \hat{\sigma}^z = \begin{pmatrix} \mathrm{e}^{-\mathrm{i}\theta/2} & 0 \\ 0 & \mathrm{e}^{\mathrm{i}\theta/2} \end{pmatrix}

$$


This is unitary and applies a phase rotation to the $|0\rangle$ state and an opposite phase to $|1\rangle$.
:::

### The Time-Dependent Schrödinger Equation

Quantum evolution is governed by a fundamental postulate: the **time-dependent Schrödinger equation**.

#### Statement

:::{admonition} Time-Dependent Schrödinger Equation
:class: important

For a quantum system with Hamiltonian $\hat{H}$ (the **energy operator**), the state $|\psi(t)\rangle$ evolves according to

$$


\mathrm{i}\hbar \frac{\partial}{\partial t} |\psi(t)\rangle = \hat{H} |\psi(t)\rangle

$$ (eq-tdse)

where $\hbar = h/(2\pi)$ is the reduced Planck constant and $\mathrm{i}$ is the imaginary unit.
:::

#### The Hamiltonian

The **Hamiltonian** $\hat{H}$ is the operator representing the **total energy** of the system:

$$


\hat{H} = \hat{T} + \hat{V}

$$


where $\hat{T}$ is kinetic energy and $\hat{V}$ is potential energy. For a single particle in 3D,

$$


\hat{H} = \frac{\hat{\boldsymbol{p}}^2}{2m} + V(\hat{\boldsymbol{r}})

$$


The Hamiltonian is **always Hermitian** ($\hat{H}^\dagger = \hat{H}$) because energy is a real, observable quantity. The eigenvalues of $\hat{H}$ are the allowed **energy levels** of the system.

:::{admonition} Discussion: Unitary Breakdown
:class: dropdown warning

In this section, we have shown that a closed quantum system evolves unitarily. But what happens when a system interacts with an environment?

- **Open system**: Suppose a qubit is coupled to a thermal bath (reservoir of other spins). The full Hamiltonian includes qubit, bath, and interaction: $\hat{H}_\text{total} = \hat{H}_\text{qubit} + \hat{H}_\text{bath} + \hat{H}_\text{int}$. The total system (qubit + bath) evolves unitarily. But we only observe the qubit. Show (conceptually or by exploring an example) that if we **trace over the bath** (sum over bath states), the reduced density matrix $\rho_\text{qubit}(t) = \text{Tr}_\text{bath}[\rho_\text{total}(t)]$ does NOT evolve unitarily. What does non-unitary evolution look like? (The general form is called a **Lindblad master equation**.)

- **Measurement collapse**: In a Stern-Gerlach experiment, a spin state is prepared as a superposition. The apparatus separates spin-up from spin-down spatially. After measuring position, the spin state appears to "collapse." Why is this not a violation of unitarity? (Hint: the apparatus is part of the system; entanglement between spin and apparatus makes the reduced spin density matrix non-unitary.)

- **Decoherence time scale**: For a macroscopic object (say, a grain of dust with $\sim 10^{18}$ atoms), the interaction with the environment is so strong that coherent superpositions become unobservable almost instantly. Is unitary evolution still happening, or has it been replaced? (Look up: "pointer basis" and "decoherence".)
:::

#### Time-Independent vs. Time-Dependent

The **time-independent Schrödinger equation** (TISE) applies when $\hat{H}$ does not depend on time:

$$


\hat{H} |\psi\rangle = E |\psi\rangle

$$


This is an eigenvalue equation for the energy eigenstates and eigenvalues. Any solution to the TISE evolves in time with a simple phase factor:

$$


|\psi(t)\rangle = \mathrm{e}^{-\mathrm{i}Et/\hbar} |\psi\rangle

$$


The **time-dependent Schrödinger equation** (TDSE), equation {eq}`eq-tdse`, is the general postulate. When $\hat{H}$ is time-independent, we can solve it and recover the TISE.

### Solution: The Time-Evolution Operator

For a **time-independent Hamiltonian**, the general solution to the TDSE can be written using the **time-evolution operator**.

#### Formal Solution

:::{admonition} Time-Evolution Operator
:class: important

The state at time $t$ is

$$


|\psi(t)\rangle = \hat{U}(t) |\psi(0)\rangle

$$ (eq-time-evolution-formula)

where the **time-evolution operator** is

$$


\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}

$$ (eq-time-evolution-operator)

Since $\hat{H}$ is Hermitian, $\hat{U}(t)$ is unitary (with the Hermitian generator being $\hat{H}/\hbar$ and the parameter being $-t$).
:::

#### Verification

:::{admonition} Derivation: Time-Evolution
:class: dropdown note

Substituting $|\psi(t)\rangle = \hat{U}(t)|\psi(0)\rangle$ into the TDSE:

$$


\mathrm{i}\hbar \frac{\partial}{\partial t} \hat{U}(t) |\psi(0)\rangle = \hat{H} \hat{U}(t) |\psi(0)\rangle

$$


Taking the time derivative of $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$:

$$


\frac{\partial \hat{U}}{\partial t} = -\frac{\mathrm{i}}{\hbar} \hat{H} \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar} = -\frac{\mathrm{i}}{\hbar} \hat{H} \hat{U}(t)

$$


Thus

$$


\mathrm{i}\hbar \frac{\partial \hat{U}}{\partial t} = \hat{H} \hat{U}(t) \quad \checkmark

$$


:::

#### Properties of the Time-Evolution Operator

The time-evolution operator satisfies several important properties:

1. **Composition law**: Evolution for time $t_1$ followed by evolution for time $t_2$ equals evolution for time $t_1 + t_2$:

$$


\hat{U}(t_1 + t_2) = \hat{U}(t_1) \hat{U}(t_2)

$$


This follows from $\mathrm{e}^{\hat{A}(t_1+t_2)} = \mathrm{e}^{\hat{A}t_1} \mathrm{e}^{\hat{A}t_2}$ for any operator $\hat{A}$.

2. **Initial condition**: $\hat{U}(0) = \mathrm{e}^0 = I$ (identity).

3. **Inverse and adjoint**: $\hat{U}^\dagger(t) = \mathrm{e}^{\mathrm{i}\hat{H}t/\hbar} = \hat{U}(-t)$. Evolution backward in time is the adjoint of forward evolution.

:::{admonition} Discussion: Hamiltonian Generator
:class: dropdown warning

For a spin-1/2 in a magnetic field along $z$: $\hat{H} = -\frac{\hbar\omega}{2}\hat{\sigma}^z$.

- **Infinitesimal evolution**: For small time step $\delta t$, the time-evolution operator is approximately $\hat{U}(\delta t) \approx \boldsymbol{I} - \frac{\mathrm{i}}{\hbar}\hat{H}\delta t + O(\delta t^2)$. (Derive this by expanding the exponential to first order.) This shows that $\hat{H}/\hbar$ is the **generator** of infinitesimal time translations.

- **Action on superposition**: Apply $\hat{U}(\delta t)$ to the state $|{+x}\rangle = \frac{1}{\sqrt{2}}(|{+z}\rangle + |{-z}\rangle)$ to first order in $\delta t$. What is the new state? Does it stay in the Bloch sphere, or does it move?

- **Rotation interpretation**: The time-evolution operator for this system is 

$$\hat{U}(t) = \mathrm{e}^{\mathrm{i}\omega t \hat{\sigma}^z/2}$$


Show that this is equivalent to a rotation by angle $\theta = \omega t$ around the $z$-axis of the Bloch sphere. What is the geometric meaning: as time evolves, the state **rotates** around the Hamiltonian direction.

- **Ehrenfest theorem connection**: In classical mechanics with Hamiltonian $H$, the equation of motion is $\frac{\mathrm{d}\boldsymbol{q}}{\mathrm{d}t} = \frac{\partial H}{\partial \boldsymbol{p}}$. In quantum mechanics, $\frac{\mathrm{d}\langle \hat{O} \rangle}{\mathrm{d}t} = \frac{\mathrm{i}}{\hbar}\langle [\hat{H}, \hat{O}]\rangle$. Do these look related? (Hint: in quantum mechanics, the commutator plays the role of the Poisson bracket.)
:::

4. **Unitarity**: $\hat{U}^\dagger(t) \hat{U}(t) = I$ for all $t$.

### Computing the Time-Evolution Operator for 2×2 Systems

For a qubit or spin-1/2 system, the Hamiltonian is a 2×2 matrix. The time-evolution operator can be computed explicitly using the spectral decomposition or by exploiting the structure of 2×2 matrices.

#### Spectral Decomposition Method

:::{admonition} Spectral: Time Evolution
:class: important

If $\hat{H}$ has eigenvalues $E_1, E_2$ and orthonormal eigenstates $|E_1\rangle, |E_2\rangle$, then

$$


\hat{H} = E_1 |E_1\rangle\langle E_1| + E_2 |E_2\rangle\langle E_2|

$$


It follows that

$$


\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar} = \mathrm{e}^{-\mathrm{i}E_1 t/\hbar} |E_1\rangle\langle E_1| + \mathrm{e}^{-\mathrm{i}E_2 t/\hbar} |E_2\rangle\langle E_2|

$$ (eq-evolution-spectral)

The eigenstates do not evolve in direction—they only accumulate phases proportional to their energies.
:::

:::{admonition} Derivation: Spectral Exponential
:class: dropdown note

For any operator $\hat{A}$ with spectral decomposition $\hat{A} = \sum_i \lambda_i |a_i\rangle\langle a_i|$, we have

$$


\mathrm{e}^{\hat{A}} = \sum_i \mathrm{e}^{\lambda_i} |a_i\rangle\langle a_i|

$$


This is because powers of $\hat{A}$ in the exponential series respect the projection structure. Applying this to $\hat{H}$:

$$


\mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar} = \sum_i \mathrm{e}^{-\mathrm{i}E_i t/\hbar} |E_i\rangle\langle E_i|

$$


:::

#### Computing $\hat{U}(t)$ for $\hat{H} = \omega \hat{\sigma}^z$

A common case is a spin in a magnetic field along the z-direction: $\hat{H} = \omega \hat{\sigma}^z = \omega \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$, where $\omega$ is a frequency.

The eigenvalues are $E_\pm = \pm \omega$ with eigenstates $|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$ and $|1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.

$$


\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\omega t} |0\rangle\langle 0| + \mathrm{e}^{\mathrm{i}\omega t} |1\rangle\langle 1| = \begin{pmatrix} \mathrm{e}^{-\mathrm{i}\omega t} & 0 \\ 0 & \mathrm{e}^{\mathrm{i}\omega t} \end{pmatrix}

$$


:::{admonition} Note
:class: note

A general state $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$ evolves to

$$


|\psi(t)\rangle = \alpha \mathrm{e}^{-\mathrm{i}\omega t} |0\rangle + \beta \mathrm{e}^{\mathrm{i}\omega t} |1\rangle

$$


Each energy eigenstate accumulates its own phase factor; the relative phase between the two evolves with the energy difference $2\omega$.
:::

### Basis Transformations and Unitary Changes of Basis

In quantum mechanics, we often switch between different bases (e.g., from the computational basis $\{|0\rangle, |1\rangle\}$ to the eigenbasis of $\hat{\sigma}^x$). These basis changes are performed by unitary transformations.

#### Change of Basis

:::{admonition} Change-of-Basis Operator
:class: important

Suppose we have an orthonormal basis $\{|e_i\rangle\}$ and another basis $\{|f_j\rangle\}$. The **change-of-basis operator** is

$$


\hat{U} = \sum_i |f_i\rangle\langle e_i|

$$ (eq-change-of-basis)

This unitary operator maps the first basis to the second: $\hat{U} |e_i\rangle = |f_i\rangle$. It is unitary because

$$


\hat{U}^\dagger \hat{U} = \sum_i |e_i\rangle\langle f_i| \sum_j |f_j\rangle\langle e_j| = \sum_i |e_i\rangle\langle e_i| = I

$$


:::

A state vector transforms as $|\psi'\rangle = \hat{U} |\psi\rangle$, and its components in the new basis are related to the old by unitary transformations of the coefficient vectors.

### Summary

- **Unitary operators** preserve inner products, normalization, and measurement probabilities. They form the foundation of reversible quantum evolution.
- **Hermitian operators** generate unitary transformations: $\hat{U} = \mathrm{e}^{\mathrm{i}\hat{G}\theta}$ where $\hat{G}^\dagger = \hat{G}$.
- The **time-dependent Schrödinger equation**, $\mathrm{i}\hbar \partial_t|\psi\rangle = \hat{H}|\psi\rangle$, is the fundamental postulate governing quantum evolution. The Hamiltonian is the energy operator.
- The **time-evolution operator** $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$ solves the Schrödinger equation. It is unitary and satisfies the composition law.
- For **time-independent Hamiltonians**, eigenstates evolve with phase factors $\mathrm{e}^{-\mathrm{i}E_n t/\hbar}$ determined by their energies. This is the basis of energy-level diagrams and selection rules.
- **Basis transformations** are unitary: any two orthonormal bases are related by a unitary operator.
- Energy eigenstates do not change in time—they only accumulate phases. Superpositions evolve, with relative phases oscillating at frequencies determined by energy differences.
