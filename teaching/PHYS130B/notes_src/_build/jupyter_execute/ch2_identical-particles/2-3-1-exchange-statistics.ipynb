## Lecture Notes

The statistics of identical particles—whether they are bosons or fermions—is encoded in how particle exchange operations act on quantum states. This section explores exchange symmetry in exotic two-dimensional systems, where particles called anyons can exhibit fractional charge and non-abelian statistics beyond the familiar boson-fermion dichotomy.

### The Key Insight: Why 2D Is Special

The existence of anyons depends on the **topology** of the configuration space of particles. The distinction between 3D and 2D is profound and has far-reaching consequences.

### Configuration Space in 3D

:::{admonition} 3D Configuration Space for Identical Particles
:class: important

For $N$ **identical** particles in 3D, the **configuration space** (positions of all particles) is:

$$


Q_N = (\mathbb{R}^3)^N / S_N

$$


where we quotient by the permutation group $S_N$ to account for indistinguishability.

**For two identical particles in 3D:**

$$


Q_2 = [\mathbb{R}^3 \times \mathbb{R}^3 \setminus \text{diagonal}] / S_2

$$


This space is **simply connected**: any closed loop can be continuously deformed to a point.
:::

:::{admonition} Derivation: Why Only Two Exchange Types in 3D
:class: dropdown

In 3D, consider two particles at positions $\boldsymbol{r}_1$ and $\boldsymbol{r}_2$. An exchange operation corresponds to a path in configuration space that continuously deforms the pair $(\boldsymbol{r}_1, \boldsymbol{r}_2)$ to $(\boldsymbol{r}_2, \boldsymbol{r}_1)$.

**Key topological fact:** Because 3D is three-dimensional, two particles can "go around" each other in an infinity of ways. For example, one particle can circle the other in the $xy$-plane, or in the $yz$-plane, or in any plane passing through both particles. 

More formally: the fundamental group $\pi_1(Q_2^{(3D)}) = \mathbb{Z}_2$. This means there are only two topologically distinct closed loops in configuration space:
- The identity (even permutation, which can be contracted to a point)
- A single exchange (odd permutation, which cannot be contracted to identity without passing through the diagonal)

Double exchange returns to identity, which forces the phase eigenvalue to satisfy $\lambda^2 = 1$, giving $\lambda = \pm 1$.
:::

**Consequence in 3D**: Only two topologically distinct paths for exchanging two particles:
- **Symmetric exchange** (even permutation): phase $+1$ (bosons)
- **Antisymmetric exchange** (odd permutation): phase $-1$ (fermions)

This binary distinction permits **only** bosons (symmetric) and fermions (antisymmetric). No intermediate statistics are possible.

### Configuration Space in 2D

:::{admonition} 2D Configuration Space and Non-Abelian Topology
:class: important

In two dimensions, the situation is dramatically different:

$$


Q_2^{(2D)} = [\mathbb{R}^2 \times \mathbb{R}^2 \setminus \text{diagonal}] / S_2

$$


**Key topological fact:** This space is **not simply connected**. It has a non-trivial fundamental group:

$$


\pi_1(Q_2^{(2D)}) = \mathbb{Z}

$$


The fundamental group is the integers, not the integers modulo 2. This means there is a continuous family of topologically distinct loops, not just two types.
:::

:::{admonition} Derivation: Why 2D Allows Continuous Exchange Angles
:class: dropdown

In 2D, all motion is confined to a plane. Consider two particles at positions $\boldsymbol{r}_1$ and $\boldsymbol{r}_2$ in the plane. An exchange path is a continuous deformation that swaps the particles.

**Key difference from 3D:** In 2D, one particle cannot "go around" the other in multiple independent ways. All exchanges happen in the same plane, and the **order** and **direction** of exchange matter — they cannot be continuously deformed into each other without passing through the diagonal (collision).

More formally: a path in $Q_2^{(2D)}$ that exchanges the two particles and returns to identity (a closed loop) winds around the diagonal. The **winding number** counts how many times the path winds around the point where the particles coincide. This winding number is a topological invariant that cannot be changed by continuous deformation.

The winding number is an integer: $n \in \mathbb{Z}$. Each winding number gives a distinct topological class of exchanges. For a single exchange operation, $n=1$. Double exchange gives $n=2$, which is distinct from $n=1$ (unlike in 3D, where double exchange returns to identity).

This continuous family of winding numbers permits a **continuous family of exchange statistics**: anyons with statistical angle $\theta = 2\pi n/m$ for integers $n, m$ or even arbitrary $\theta \in [0, 2\pi)$.
:::

### Discussion: Is Topology a Reason or Just a Constraint?

:::{admonition} Discussion
:class: dropdown warning

**Configuration Space Topology and Allowed Spin Values:** For $N$ identical particles in 3D, the configuration space is $Q_N = (\mathbb{R}^3)^N / S_N$ (the permutation group quotient acts trivially on each particle's 3D position). The fundamental group is $\pi_1(Q_N) \cong S_N$ for $N \geq 2$, which is a **discrete non-abelian group**.

In 2D, however, $Q_N = (\mathbb{R}^2)^N / S_N$ has $\pi_1(Q_N) \cong B_N$ (the braid group), which is **infinite and non-abelian**.

**Computational questions:**
1. For 3D with two particles, the two-fold adiabatic loops (path exchanges) form the group $\mathbb{Z}_2$. Any representation of a loop exchange must give a unitary $U$ with $U^2 = \pm \mathbb{1}$. Show that $U^2 = 1$ (trivial) gives integer-spin bosons, and $U^2 = -\mathbb{1}$ gives half-integer-spin fermions, but **no other phases are allowed**.

2. In 2D, the braid group allows many more representations. Show that you can have $U^N = 1$ for any integer $N$, corresponding to anyons with fractional statistical angle $\theta = 2\pi/N$.

3. Why does the 2D configuration space topology (braid group vs permutation group) directly permit anyonic statistics, while 3D topology forbids it?
:::

**Consequence in 2D**: Continuous families of exchange statistics are possible—anyons.

### The Fundamental Exchange Operator

:::{admonition} Exchange Operator and Eigenvalues
:class: important

The **exchange operator** $P_{12}$ acts on two-particle states by swapping the particle labels:

$$


P_{12} |\psi(\boldsymbol{r}_1, \boldsymbol{r}_2)\rangle = |\psi(\boldsymbol{r}_2, \boldsymbol{r}_1)\rangle

$$


For identical particles, the wavefunction must be an eigenstate of $P_{12}$:

$$


P_{12} |\psi\rangle = \lambda |\psi\rangle

$$


where $\lambda$ is the exchange eigenvalue.

**In 3D:** $\lambda = \pm 1$
- $\lambda = +1$ → bosons (symmetric wavefunction)
- $\lambda = -1$ → fermions (antisymmetric wavefunction)

**In 2D:** $\lambda = \mathrm{e}^{\mathrm{i}\theta}$ for any $\theta \in [0, 2\pi)$ → anyons
:::

:::{admonition} Two-Particle Wavefunction Symmetry
:class: important

For identical particles:
- **Bosons:** $\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = +\psi(\boldsymbol{r}_2, \boldsymbol{r}_1)$ (symmetric)
- **Fermions:** $\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = -\psi(\boldsymbol{r}_2, \boldsymbol{r}_1)$ (antisymmetric)
- **Anyons:** $\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = \mathrm{e}^{\mathrm{i}\theta}\psi(\boldsymbol{r}_2, \boldsymbol{r}_1)$ (fractional phase)

This is a **global constraint**, not a detail—it rules out half of the naïve tensor product space and has dramatic consequences for thermodynamics and quantum information.
:::

### Discussion: Fermi Holes and the Physics of Exchange Symmetry

:::{admonition} Discussion
:class: dropdown warning

**Fermi Holes and Spatial Separation:** Consider two identical fermions (both spin-down) in a box, in the ground state and first excited state. The symmetric part of the spatial wavefunction is:

$$


\Psi_{\text{sym}}(\boldsymbol{r}_1, \boldsymbol{r}_2) = \frac{1}{\sqrt{2}}[\psi_1(\boldsymbol{r}_1)\psi_2(\boldsymbol{r}_2) + \psi_1(\boldsymbol{r}_2)\psi_2(\boldsymbol{r}_1)]

$$


The **antisymmetric** wavefunction (required for fermions) is:

$$


\Psi_{\text{antisym}}(\boldsymbol{r}_1, \boldsymbol{r}_2) = \frac{1}{\sqrt{2}}[\psi_1(\boldsymbol{r}_1)\psi_2(\boldsymbol{r}_2) - \psi_1(\boldsymbol{r}_2)\psi_2(\boldsymbol{r}_1)]

$$


**Question:** Compute the probability density $|\Psi_{\text{antisym}}(\boldsymbol{r}_1, \boldsymbol{r}_2)|^2$ at the point $\boldsymbol{r}_1 = \boldsymbol{r}_2 = \boldsymbol{r}$. Show that it vanishes identically. Then compute the same for bosons (symmetric state) and for two distinguishable particles (product state). What does this calculation reveal about the physical origin of the **Fermi hole**—the depletion of two-particle density near the diagonal?
:::

### Physical Consequences

:::{admonition} Wavefunction Form Restrictions
:class: caution

For **two fermions** with the same spin: The wavefunction must vanish if $\boldsymbol{r}_1 = \boldsymbol{r}_2$ (Pauli exclusion principle). This prevents two fermions from occupying the same spatial state with the same spin.

For **two bosons**: The wavefunction can be nonzero at $\boldsymbol{r}_1 = \boldsymbol{r}_2$, which allows macroscopic occupation of a single state—**Bose condensation**. This underlies superfluidity and superconductivity.

For **anyons** in 2D: Exchange statistics are modified but not restrictive in the same way. Anyons can have fractional statistics and fractional charge, with applications to the fractional quantum Hall effect.
:::

### Discussion: The Gibbs Paradox in Quantum Mechanics

:::{admonition} Discussion
:class: dropdown warning

**The Gibbs Paradox and the $N!$ Factor:** In classical mechanics, the phase space volume for $N$ indistinguishable particles is:

$$


V_{\text{indist}} = \frac{1}{N!}\int \mathrm{d}^{3N}p \, \mathrm{d}^{3N}q \, f(E)

$$


where $N!$ accounts for the fact that permutations of identical particles represent the same microstate. In quantum mechanics, this translates directly: the number of distinguishable microstates is reduced by a factor $N!$ compared to a naive count.

**Question:** Two isolated containers each hold $N$ identical fermions at the same temperature $T$. Initially, partition function for each is $Z_1 = Z_2 = Z_0$ (unknown). When the containers are **allowed to mix** (removing the partition), we must recalculate the total system partition function:

$$


Z_{\text{before}} = Z_0 \cdot Z_0 = Z_0^2

$$


$$

Z_{\text{after}} = \frac{1}{(2N)!}\int \mathrm{d}^{3(2N)}p \, \mathrm{d}^{3(2N)}q \, f(E)

$$


Show that the entropy change is:

$$


\Delta S = k_B \ln\frac{Z_{\text{after}}}{Z_{\text{before}}} = -k_B \ln[(2N)!/(\, 2 \cdot N! \cdot N!\,)]

$$


Using Stirling's approximation, show this is **negative** (entropy decreases) for large $N$. Explain: why does mixing identical fermions reduce entropy, contrary to the naive expectation? Why is this impossible for distinguishable particles?
:::

### Topology Matters

:::{admonition} Key Topological Result
:class: note

In 3D, exchanging two particles twice is topologically trivial (contractible to identity). This forces the exchange phase to satisfy $\mathrm{e}^{2\mathrm{i}\theta} = 1$, giving only $\theta \in \{0, \pi\}$ (bosons or fermions).

In 2D, double exchange is **non-trivial**. It corresponds to a winding number of 2 in configuration space, which is distinct from winding number 1 (single exchange). This permits continuous $\theta$ and anyons.

The difference is purely topological: the dimensionality of space determines the fundamental group of configuration space, which determines what exchange statistics are possible.
:::

### Summary

**Key Takeaways:**

1. **Configuration space topology determines exchange statistics.** In 3D, $\pi_1(Q_2) = \mathbb{Z}_2$ forces only bosons or fermions. In 2D, $\pi_1(Q_2) = \mathbb{Z}$ permits a continuous family of anyons.

2. **The exchange operator eigenvalue encodes statistics.** Bosons have eigenvalue $+1$, fermions $-1$, and anyons have eigenvalue $\mathrm{e}^{\mathrm{i}\theta}$ with arbitrary $\theta$.

3. **Dimensionality is fundamental.** This is not a subtle relativistic effect but a purely topological consequence of space dimension. A 2D system confined to a plane can support qualitatively different particle statistics than a 3D system.

4. **Indistinguishability is built in.** By quotienting by $S_N$, we impose that particles are identical. This simple mathematical step has profound consequences: it divides the Hilbert space in half and determines whether fermions exhibit Pauli exclusion or bosons exhibit condensation.
## The Braid Group vs. Permutation Group

### Permutation Group $S_N$ (3D)

In three dimensions, only the **topology** of the exchange matters, not its specific trajectory. Thus, particle exchange is described by the **permutation group** $S_N$:

- An even permutation has phase $+1$.
- An odd permutation has phase $-1$.
- Permutations **compose abelianly**: the order of exchanges doesn't matter (up to their parity).

### The Braid Group $B_N$ (2D)

In two dimensions, we must track **how** particles exchange, not just **whether** they do. This is encoded by the **braid group** $B_N$.

**Definition**: A braid is a sequence of particle exchanges (strands) in 2D, where:
- Each exchange involves crossing two particles.
- Crossings can be clockwise or counterclockwise.
- The **order** of exchanges matters: composing two braids gives a different braid if done in different order.

For two particles, the braid group $B_2$ is generated by a single element $\sigma$ (exchange), which satisfies:

$$


\hat{\sigma}^2 \neq e \quad \text{(unlike the permutation group, where } \hat{\sigma}^2 = e\text{)}

$$


Instead, only repeated application to a high power gives the identity.

**Non-abelian structure**: For $N \geq 3$, the braid group is **non-abelian**: $\hat{\sigma}_i \hat{\sigma}_j \neq \hat{\sigma}_j \hat{\sigma}_i$ for $i \not\in \{j-1, j, j+1\}$.

### Yang-Baxter Equation

The consistency condition for braiding is the **Yang-Baxter equation**:

$$


R_{12} R_{23} R_{12} = R_{23} R_{12} R_{23}

$$


where $R_{ij}$ is the braiding operation on particles $i$ and $j$. This ensures that different braiding sequences are compatible—no contradictions arise from different orderings.
## Exchange Statistics in 2D: Anyons

### The Adiabatic Phase

When two identical particles **exchange** adiabatically in 2D, the wavefunction can acquire an arbitrary **phase factor**:

$$


\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) \to \mathrm{e}^{\mathrm{i}\phi} \psi(\boldsymbol{r}_2, \boldsymbol{r}_1)

$$ (eq-anyon-exchange-phase)

where $\phi$ can be **any real number** (or often, quantized in units of $2\pi$).

### The Statistical Angle

The exchange phase is often written as:

$$


\mathrm{e}^{\mathrm{i}\pi\theta}

$$


where $\theta$ is the **statistical angle** or **anyon angle**, expressed in units of $2\pi$ (i.e., $\theta \in [0, 2)$).

$$


\boxed{\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = \mathrm{e}^{\mathrm{i}\pi\theta} \psi(\boldsymbol{r}_2, \boldsymbol{r}_1)}

$$ (eq-anyon-statistics)

### Recovering Bosons and Fermions

**Bosons**: $\theta = 0 \Rightarrow \mathrm{e}^{\mathrm{i}\pi \cdot 0} = +1$ (symmetric).

**Fermions**: $\theta = 1 \Rightarrow \mathrm{e}^{\mathrm{i}\pi \cdot 1} = -1$ (antisymmetric).

**Fractional anyons**: $\theta = 1/3 \Rightarrow \mathrm{e}^{\mathrm{i}\pi/3}$ (phase of $60°$).

Any value between 0 and 2 is allowed (in principle), though physical systems often have specific quantized values.
## Braiding Operations and Quantum Dimension

### Braiding as a Unitary Operator

In a system with $N$ anyons, the **braiding operation** that exchanges particles $i$ and $j$ is a unitary operator $R_{ij}$ in Hilbert space. When applied to a many-anyon state:

$$


R_{ij} |\psi\rangle = \mathrm{e}^{\mathrm{i}\pi\theta} |\psi'\rangle

$$


For **non-abelian anyons**, braiding can mix different states—it's not just a phase factor.

### Quantum Dimension

One of the most striking features of anyons is the concept of **quantum dimension**. For a system of $N$ identical anyons:

$$


\dim(\mathcal{H}_N) \sim d^N

$$


where $d$ is the **quantum dimension** of the anyon type.

- For bosons and fermions in 3D: $d = 1$ (all $N$-particle states reduce to a single Fock sector).
- For **non-abelian Fibonacci anyons** (found in certain quantum Hall states): $d = \phi = \frac{1 + \sqrt{5}}{2} \approx 1.618$ (the golden ratio!).

**Consequence**: The Hilbert space dimension grows **exponentially** with $N$, but the growth rate is $\phi^N$ instead of $2^N$.

$$


d^N = \phi^N \approx 1.618^N

$$


This provides exponential advantage for quantum computation if anyons can be manipulated and measured.
## One and Two Dimensions: Comparison

### One Dimension

In 1D, particles on a line **cannot pass through each other**. "Exchange" is not well-defined—if two particles try to swap positions, they must collide.

**Consequence**: Identical particles in 1D don't fall into universal boson/fermion classes. Instead, they are described by **Tomonaga-Luttinger liquid theory**, which shows power-law correlations rather than simple permutation statistics.

### Two Dimensions

The configuration space has non-trivial topology $\Rightarrow$ **anyonic statistics** possible.

### Three Dimensions and Higher

The configuration space is simply connected $\Rightarrow$ only bosons and fermions.

**Theorem**: In dimensions $d \geq 3$, only two types of statistics (boson and fermion) are allowed.
:::{admonition} Example: Examples of 2D Systems
:class: example


### Quantum Hall Systems

A 2D electron gas in a strong perpendicular magnetic field forms a **quantum Hall state**. Quasiparticles (excitations) in these states are anyons!

### Cold Atoms in Optical Lattices

Atoms trapped in a 2D optical lattice (a periodic potential created by laser beams) can be engineered to have anyonic statistics via **synthetic gauge fields**.

### Molecular Systems

Some proposals involve engineered molecules or superconductors to realize 2D anyonic systems.

### Topological Insulators

The edge states of 2D topological insulators can host anyonic quasiparticles.
:::
## The Fundamental Reason: Homotopy of Configuration Space

### Fundamental Group

The rigorous statement involves the **fundamental group** $\pi_1(Q_N)$ of the configuration space.

- **3D**: $\pi_1(Q_N^{(3D)}) = S_N$ (permutation group) for $N \geq 2$.
- **2D**: $\pi_1(Q_N^{(2D)}) = B_N$ (braid group) for $N \geq 2$.
- **1D**: Configuration space is not even well-defined (particles can't pass); no ordinary exchange.

### Representations of Braid Groups

Since the wavefunction is a **representation** of the fundamental group, and $B_N$ has vastly more representations than $S_N$, we get vastly more statistics in 2D.

- Each representation of $B_N$ corresponds to a possible set of anyonic statistics.
- Abelian representations (1D) give abelian anyons (fractional statistics).
- Non-abelian representations give non-abelian anyons (richer structure).
## Summary

- **Configuration space topology**: In 3D, it's simply connected → only bosons/fermions. In 2D, it's not → anyons possible.
- **Braid group** $B_N$ (2D) replaces permutation group $S_N$ (3D); braiding order matters.
- **Anyonic statistics**: Exchange phase $\mathrm{e}^{\mathrm{i}\pi\theta}$ with arbitrary $\theta \in [0, 2)$.
- **Bosons** ($\theta = 0$) and **fermions** ($\theta = 1$) are special cases.
- **Quantum dimension** $d > 1$ for non-abelian anyons → exponential growth of Hilbert space.
- **Physical realizations**: Quantum Hall systems, cold atoms, topological materials.
