## Lecture Notes

The statistics of identical particles—whether they are bosons or fermions—is encoded in how particle exchange operations act on quantum states. This section explores exchange symmetry in exotic two-dimensional systems, where particles called anyons can exhibit fractional charge and non-abelian statistics beyond the familiar boson-fermion dichotomy.
### The Key Insight: Why 2D Is Special

The existence of anyons depends on the **topology** of the configuration space of particles. Let us contrast 3D and 2D.

### Configuration Space in 3D

For $N$ **identical** particles in 3D, the **configuration space** (positions of all particles) is:

$$

Q_N = (\mathbb{R}^3)^N / S_N

$$

where we quotient by the permutation group $S_N$ to account for indistinguishability.

**Key topological fact**: The configuration space of two identical particles in 3D, with the diagonal removed (i.e., excluding coincident positions), is:

$$

Q_2 = [\mathbb{R}^3 \times \mathbb{R}^3 \setminus \text{diagonal}] / S_2

$$

This space is **simply connected**: any closed loop can be continuously deformed to a point.

**Consequence in 3D**: Only two topologically distinct paths for exchanging two particles:
- **Even exchange**: adiabatically deform into an even permutation.
- **Odd exchange**: adiabatically deform into an odd permutation.

This binary distinction permits **only** bosons (symmetric, phase $+1$) and fermions (antisymmetric, phase $-1$).

### Configuration Space in 2D

In two dimensions, the situation is dramatically different:

$$

Q_2^{(2D)} = [\mathbb{R}^2 \times \mathbb{R}^2 \setminus \text{diagonal}] / S_2

$$

**Key topological fact**: This space is **not simply connected**. It has a non-trivial fundamental group.

**Why?** In 3D, two particles can "go around" each other in an infinity of ways (circling in any plane). In 2D, all exchanges must happen in the same plane, and the **order** and **direction** of exchange matter—they cannot be continuously deformed into each other.

**Consequence in 2D**: Continuous families of exchange statistics are possible—anyons.

:::{admonition} Topology matters
:class: note
In 3D, exchanging two particles twice is topologically trivial (contractible to identity), forcing the exchange phase to satisfy $e^{2i\theta} = 1$, giving only bosons ($\theta=0$) or fermions ($\theta=\pi$). In 2D, double exchange is non-trivial, allowing continuous $\theta$—anyons.
:::

:::{important} Exchange Operator and Eigenvalues
:class: important

The exchange operator P₁₂ swaps two particles. For identical particles, P₁₂|ψ⟩ = λ|ψ⟩ with λ = ±1. Eigenvalue +1 (bosons) gives symmetric states; -1 (fermions) gives antisymmetric states.
:::

:::{note} Two-Particle Wavefunction Symmetry
:class: note

For identical particles: ψ(r₁,r₂) = +ψ(r₂,r₁) for bosons; ψ(r₁,r₂) = -ψ(r₂,r₁) for fermions. This is a global constraint, not a detail—it rules out half of the naïve tensor product space.
:::

:::{caution} Wavefunction Form Restrictions
:class: caution

For two fermions, the wavefunction must vanish if r₁=r₂ (cannot occupy same position with same spin). For two bosons, wavefunction can be nonzero at r₁=r₂, which allows Bose condensation phenomena.
:::


## The Braid Group vs. Permutation Group

### Permutation Group $S_N$ (3D)

In three dimensions, only the **topology** of the exchange matters, not its specific trajectory. Thus, particle exchange is described by the **permutation group** $S_N$:

- An even permutation has phase $+1$.
- An odd permutation has phase $-1$.
- Permutations **compose abelianly**: the order of exchanges doesn't matter (up to their parity).

### The Braid Group $B_N$ (2D)

In two dimensions, we must track **how** particles exchange, not just **whether** they do. This is encoded by the **braid group** $B_N$.

**Definition**: A braid is a sequence of particle exchanges (strands) in 2D, where:
- Each exchange involves crossing two particles.
- Crossings can be clockwise or counterclockwise.
- The **order** of exchanges matters: composing two braids gives a different braid if done in different order.

For two particles, the braid group $B_2$ is generated by a single element $\sigma$ (exchange), which satisfies:

$$

\hat{\sigma}^2 \neq e \quad \text{(unlike the permutation group, where } \hat{\sigma}^2 = e\text{)}

$$

Instead, only repeated application to a high power gives the identity.

**Non-abelian structure**: For $N \geq 3$, the braid group is **non-abelian**: $\hat{\sigma}_i \hat{\sigma}_j \neq \hat{\sigma}_j \hat{\sigma}_i$ for $i \not\in \{j-1, j, j+1\}$.

### Yang-Baxter Equation

The consistency condition for braiding is the **Yang-Baxter equation**:

$$

R_{12} R_{23} R_{12} = R_{23} R_{12} R_{23}

$$

where $R_{ij}$ is the braiding operation on particles $i$ and $j$. This ensures that different braiding sequences are compatible—no contradictions arise from different orderings.

## Exchange Statistics in 2D: Anyons

### The Adiabatic Phase

When two identical particles **exchange** adiabatically in 2D, the wavefunction can acquire an arbitrary **phase factor**:

$$

\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) \to e^{i\phi} \psi(\boldsymbol{r}_2, \boldsymbol{r}_1)

$$ (eq-anyon-exchange-phase)

where $\phi$ can be **any real number** (or often, quantized in units of $2\pi$).

### The Statistical Angle

The exchange phase is often written as:

$$

e^{i\pi\theta}

$$

where $\theta$ is the **statistical angle** or **anyon angle**, expressed in units of $2\pi$ (i.e., $\theta \in [0, 2)$).

$$

\boxed{\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = e^{i\pi\theta} \psi(\boldsymbol{r}_2, \boldsymbol{r}_1)}

$$ (eq-anyon-statistics)

### Recovering Bosons and Fermions

**Bosons**: $\theta = 0 \Rightarrow e^{i\pi \cdot 0} = +1$ (symmetric).

**Fermions**: $\theta = 1 \Rightarrow e^{i\pi \cdot 1} = -1$ (antisymmetric).

**Fractional anyons**: $\theta = 1/3 \Rightarrow e^{i\pi/3}$ (phase of $60°$).

Any value between 0 and 2 is allowed (in principle), though physical systems often have specific quantized values.

## Braiding Operations and Quantum Dimension

### Braiding as a Unitary Operator

In a system with $N$ anyons, the **braiding operation** that exchanges particles $i$ and $j$ is a unitary operator $R_{ij}$ in Hilbert space. When applied to a many-anyon state:

$$

R_{ij} |\psi\rangle = e^{i\pi\theta} |\psi'\rangle

$$

For **non-abelian anyons**, braiding can mix different states—it's not just a phase factor.

### Quantum Dimension

One of the most striking features of anyons is the concept of **quantum dimension**. For a system of $N$ identical anyons:

$$

\dim(\mathcal{H}_N) \sim d^N

$$

where $d$ is the **quantum dimension** of the anyon type.

- For bosons and fermions in 3D: $d = 1$ (all $N$-particle states reduce to a single Fock sector).
- For **non-abelian Fibonacci anyons** (found in certain quantum Hall states): $d = \phi = \frac{1 + \sqrt{5}}{2} \approx 1.618$ (the golden ratio!).

**Consequence**: The Hilbert space dimension grows **exponentially** with $N$, but the growth rate is $\phi^N$ instead of $2^N$.

$$

d^N = \phi^N \approx 1.618^N

$$

This provides exponential advantage for quantum computation if anyons can be manipulated and measured.

## One and Two Dimensions: Comparison

### One Dimension

In 1D, particles on a line **cannot pass through each other**. "Exchange" is not well-defined—if two particles try to swap positions, they must collide.

**Consequence**: Identical particles in 1D don't fall into universal boson/fermion classes. Instead, they are described by **Tomonaga-Luttinger liquid theory**, which shows power-law correlations rather than simple permutation statistics.

### Two Dimensions

The configuration space has non-trivial topology $\Rightarrow$ **anyonic statistics** possible.

### Three Dimensions and Higher

The configuration space is simply connected $\Rightarrow$ only bosons and fermions.

**Theorem**: In dimensions $d \geq 3$, only two types of statistics (boson and fermion) are allowed.

:::{admonition} Example: Examples of 2D Systems
:class: example


### Quantum Hall Systems

A 2D electron gas in a strong perpendicular magnetic field forms a **quantum Hall state**. Quasiparticles (excitations) in these states are anyons!

### Cold Atoms in Optical Lattices

Atoms trapped in a 2D optical lattice (a periodic potential created by laser beams) can be engineered to have anyonic statistics via **synthetic gauge fields**.

### Molecular Systems

Some proposals involve engineered molecules or superconductors to realize 2D anyonic systems.

### Topological Insulators

The edge states of 2D topological insulators can host anyonic quasiparticles.
:::

## The Fundamental Reason: Homotopy of Configuration Space

### Fundamental Group

The rigorous statement involves the **fundamental group** $\pi_1(Q_N)$ of the configuration space.

- **3D**: $\pi_1(Q_N^{(3D)}) = S_N$ (permutation group) for $N \geq 2$.
- **2D**: $\pi_1(Q_N^{(2D)}) = B_N$ (braid group) for $N \geq 2$.
- **1D**: Configuration space is not even well-defined (particles can't pass); no ordinary exchange.

### Representations of Braid Groups

Since the wavefunction is a **representation** of the fundamental group, and $B_N$ has vastly more representations than $S_N$, we get vastly more statistics in 2D.

- Each representation of $B_N$ corresponds to a possible set of anyonic statistics.
- Abelian representations (1D) give abelian anyons (fractional statistics).
- Non-abelian representations give non-abelian anyons (richer structure).

## Summary

- **Configuration space topology**: In 3D, it's simply connected → only bosons/fermions. In 2D, it's not → anyons possible.
- **Braid group** $B_N$ (2D) replaces permutation group $S_N$ (3D); braiding order matters.
- **Anyonic statistics**: Exchange phase $e^{i\pi\theta}$ with arbitrary $\theta \in [0, 2)$.
- **Bosons** ($\theta = 0$) and **fermions** ($\theta = 1$) are special cases.
- **Quantum dimension** $d > 1$ for non-abelian anyons → exponential growth of Hilbert space.
- **Physical realizations**: Quantum Hall systems, cold atoms, topological materials.
