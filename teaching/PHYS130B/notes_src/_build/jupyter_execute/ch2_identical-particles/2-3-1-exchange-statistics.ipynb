## Lecture Notes

### Overview

In 3D, identical particles must be bosons or fermions — these are the only two options allowed by quantum mechanics. But in 2D, the topology of particle worldlines is richer: exchanging two particles twice is **not** topologically trivial, so any exchange phase is allowed. This lesson explores why 2D topology opens the door to **anyons** and develops the charge-flux composite as a concrete toy model for anyonic statistics.

### Why 2D Is Special

:::{margin}
:::{figure} ../assets/figs/2-3-1-exchange-topology.png
:name: fig-exchange-topology-2d-vs-3d
:width: 100%

Exchange topology: in 2D, clockwise and counterclockwise exchanges are distinct ($\hat{\mathcal{P}} \neq \hat{\mathcal{P}}^{-1}$); in 3D they are continuously deformable into each other and become equivalent.
:::
:::

In 3D, exchanging two identical particles twice is the same as doing nothing — the particles can always "go around" each other and return to the original configuration. This forces the exchange phase to satisfy $\mathrm{e}^{2\mathrm{i}\theta} = 1$, giving only two possibilities:

:::{admonition} Exchange Statistics in 3D
:class: important

$$\text{Bosons: } \theta = 0 \quad (\text{no phase}), \qquad \text{Fermions: } \theta = \pi \quad (\text{phase } {-1})$$

These are the only options in three spatial dimensions.
:::
:::{admonition} Proof: Why Only $\theta = 0$ or $\pi$ in 3D
:class: dropdown information

The allowed exchange phases are determined by the **topology of configuration space**. For $N$ identical particles in $d$ dimensions, the configuration space is

$$
\mathcal{C}_N = (\mathbb{R}^{dN} \setminus \Delta) / S_N
$$

where $\Delta$ removes coincident points (particles cannot overlap) and $S_N$ quotients by permutations (particles are identical). Quantum statistics is classified by the **first fundamental group** $\pi_1(\mathcal{C}_N)$: each topologically distinct loop in configuration space corresponds to a distinct exchange process, and the wavefunction must pick up a well-defined phase for each.

**In 3D ($d \geq 3$):** The fundamental group is the permutation group, $\pi_1(\mathcal{C}_N) = S_N$. The key fact is that for $d \geq 3$, any loop where particle 1 goes around particle 2 can be continuously deformed to the trivial loop — the extra spatial dimensions provide room to "lift" the path over the obstacle. Since the exchange operation $\hat{\mathcal{P}}_{12}$ satisfies $\hat{\mathcal{P}}_{12}^2 = I$ in $S_N$, the phase must satisfy

$$
\mathrm{e}^{2\mathrm{i}\theta} = 1 \quad \Longrightarrow \quad \theta = 0 \text{ (boson) or } \theta = \pi \text{ (fermion)}
$$

**In 2D ($d = 2$):** The fundamental group is the braid group, $\pi_1(\mathcal{C}_N) = B_N$. In a plane, one particle cannot "go over" another — the path around the obstacle is topologically nontrivial and cannot be shrunk to a point. The braid group has generators $\hat{\mathcal{P}}_i$ that satisfy $\hat{\mathcal{P}}_i \hat{\mathcal{P}}_{i+1} \hat{\mathcal{P}}_i = \hat{\mathcal{P}}_{i+1} \hat{\mathcal{P}}_i \hat{\mathcal{P}}_{i+1}$ (the Yang-Baxter equation) but **not** $\hat{\mathcal{P}}_i^2 = I$. The one-dimensional representations of $B_N$ are $\hat{\mathcal{P}}_i \mapsto \mathrm{e}^{\mathrm{i}\theta}$ for any $\theta \in [0, 2\pi)$, giving abelian anyons. Higher-dimensional representations give non-abelian anyons.
:::


In **2D**, the situation changes. When one particle moves around another in a plane, it traces out a loop that **cannot be shrunk to a point** — the other particle is always "in the way." Clockwise and counterclockwise exchanges are topologically distinct, so in general $\hat{\mathcal{P}} \neq \hat{\mathcal{P}}^{-1}$. Exchanging twice is no longer trivial. The exchange phase $\mathrm{e}^{\mathrm{i}\theta}$ can take **any value**, not just $\pm 1$.


:::{admonition} Braiding vs. Permutation
:class: note

In 3D, particle exchange is described by the **permutation group** $S_N$. Exchanging twice gives the identity: $\hat{\mathcal{P}}_{12}^2 = I$.

In 2D, particle exchange is described by the **braid group** $B_N$. Exchanging twice does **not** give the identity — the worldlines of the particles form a braid that cannot be untangled.
:::

:::{margin}
:::{figure} ../assets/figs/2-3-1-braid-group.png
:width: 100%
:name: fig-braid-group

Braid group: two elementary exchanges $\hat{\mathcal{P}}_{12}$ and $\hat{\mathcal{P}}_{23}$ acting on three strands.
:::
:::

Think of particle worldlines as strings extending in time. In 2D+1D spacetime:

- Two strings can wind around each other, forming knots and braids
- The order in which you braid matters (for non-abelian anyons)
- The number of windings is a topological invariant — it cannot be changed by smooth deformations


:::{admonition} Definition: (Abelian) Anyons
:class: important

Particles in 2D with exchange phase $\theta \neq 0, \pi$ are called **anyons** (from "any-ons," because $\theta$ can be *any* value). Under a single counterclockwise exchange, the many-body state $\vert\Psi\rangle$ acquires a phase:

$$|\Psi\rangle \to \mathrm{e}^{\mathrm{i}\theta} |\Psi\rangle$$

- $\theta = 0$: boson
- $\theta = \pi$: fermion  
- $0 < \theta < \pi$: anyon (fractional statistics)
:::

:::{admonition} Abelian vs. Non-Abelian
:class: note

If exchanging particles only produces a **phase** $\mathrm{e}^{\mathrm{i}\theta}$, the anyons are **abelian** — different exchanges commute.

If exchanging particles acts as a **unitary matrix** on a degenerate ground-state space, i.e. $|\Psi\rangle \to U |\Psi\rangle$, the anyons are **non-abelian** — the order of exchanges matters.
:::

### Charge-Flux Composite

A useful toy model for abelian anyons is the **charge-flux composite** $(q, \Phi)$: a point charge $q$ bound to a thin magnetic flux tube $\Phi$. The Aharonov-Bohm effect then generates statistical phases.

:::{admonition} Aharonov-Bohm Phase
:class: important

When a charge $q$ completes a closed loop encircling a magnetic flux $\Phi$, it accumulates a phase:

$$

\theta_{\mathrm{AB}} = \frac{q\,\Phi}{\hbar}

$$ (eq-ab-phase)
:::

:::{admonition} Derivation: Aharonov–Bohm Phase from the Path Integral
:class: dropdown information

We derive the Aharonov–Bohm phase accumulated by a charged particle moving along a closed loop $\mathcal{C}$ encircling a magnetic flux tube carrying flux $\Phi$.

**Setup.** Outside the flux tube the magnetic field vanishes, $\boldsymbol{B} = 0$, but the vector potential $\boldsymbol{A}$ is non-zero and satisfies

$$
\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{r} = \Phi
$$

by Stokes' theorem applied to any surface bounded by $\mathcal{C}$.

**Minimal coupling in the Lagrangian.** For a particle of charge $q$ moving in an electromagnetic field, the Lagrangian contains the minimal-coupling term:

$$
L = \frac{1}{2}m\dot{\boldsymbol{r}}^2 + \frac{q}{c}\boldsymbol{A}(\boldsymbol{r}) \cdot \dot{\boldsymbol{r}} \qquad (\text{Gaussian units})
$$

or equivalently in SI units $q\boldsymbol{A}\cdot\dot{\boldsymbol{r}}$.

**Path integral phase.** In the path integral, each path $\boldsymbol{r}(t)$ contributes a weight $\exp(\mathrm{i}S[\boldsymbol{r}]/\hbar)$. The electromagnetic contribution to the action along a path is

$$
S_{\mathrm{em}} = q\int \boldsymbol{A}(\boldsymbol{r}(t))\cdot\dot{\boldsymbol{r}}\,\mathrm{d}t = q\int_{\mathcal{C}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{r}
$$

This is purely a line integral along the path and does not depend on the speed of travel.

**Closed loop.** For a closed loop $\mathcal{C}$ encircling the flux tube, all paths that wind around the tube once receive the same additional phase factor regardless of their detailed shape:

$$
\mathrm{e}^{\mathrm{i}S_{\mathrm{em}}/\hbar} = \exp\!\left(\frac{\mathrm{i}q}{\hbar}\oint_{\mathcal{C}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{r}\right) = \exp\!\left(\frac{\mathrm{i}q\Phi}{\hbar}\right)
$$

**Physical consequence.** Even though $\boldsymbol{B} = 0$ along every path, the vector potential $\boldsymbol{A}$ is non-zero and gives a gauge-invariant (observable) phase when integrated around the loop. This phase is $\theta_{\mathrm{AB}} = q\Phi/\hbar$, independent of the loop's size or shape as long as it winds once around the flux tube.
:::

This single formula unifies three kinds of statistics for charge-flux composites:

:::{admonition} Three Kinds of Statistics
:class: important

- **Braiding statistics.** When one composite $(q_1, \Phi_1)$ moves in a full closed loop around another $(q_2, \Phi_2)$, the charge $q_1$ encircles the flux $\Phi_2$, accumulating an Aharonov-Bohm phase:

    $$
    \theta_{\mathrm{br}} = \frac{q_1 \Phi_2}{\hbar}
    $$ (eq-braid-phase)

- **Exchange statistics.** Exchanging two identical composites $(q, \Phi)$ is a half-braid, so the exchange phase is:

    $$
    \theta_{\mathrm{ex}} = \frac{\theta_{\mathrm{br}}}{2} = \frac{q\,\Phi}{2\hbar}
    $$ (eq-exchange-phase)

- **Spin-statistics relation.** A $2\pi$ self-rotation of a composite produces the same phase as an exchange. The spin quantum number is:

    $$
    s = \frac{\theta_{\mathrm{ex}}}{2\pi} = \frac{q\,\Phi}{4\pi\hbar}
    $$ (eq-spin-phase)
:::

:::{admonition} Derivation: Exchange Phase Is Half the Braiding Phase
:class: dropdown information

Consider two identical charge-flux composites $(q, \Phi)$ at positions $\boldsymbol{r}_1$ and $\boldsymbol{r}_2$. We want to show that exchanging them (a half-braid) gives exactly half the phase of a full braid.

**Full braid (braiding phase):** Composite 1 makes a complete counterclockwise loop around composite 2, returning to $\boldsymbol{r}_1$. The charge $q_1$ encircles the flux $\Phi_2$, accumulating the Aharonov-Bohm phase $\theta_{\mathrm{br}} = q\Phi/\hbar$.

**Half-braid (exchange phase):** In an exchange, composite 1 moves to position $\boldsymbol{r}_2$ while composite 2 simultaneously moves to $\boldsymbol{r}_1$, with each tracing a semicircular arc. In the **center-of-mass frame**, the relative coordinate $\boldsymbol{r} = \boldsymbol{r}_1 - \boldsymbol{r}_2$ traces a full semicircle (angle $\pi$), so the relative angle changes by $\Delta\phi = \pi$.

But a full braid corresponds to $\Delta\phi = 2\pi$ in the relative coordinate. Since the Aharonov-Bohm phase is proportional to the subtended angle:

$$
\theta_{\mathrm{ex}} = \frac{\Delta\phi_{\text{exchange}}}{\Delta\phi_{\text{braid}}} \cdot \theta_{\mathrm{br}} = \frac{\pi}{2\pi} \cdot \frac{q\Phi}{\hbar} = \frac{q\Phi}{2\hbar}
$$

Equivalently, the charge $q_1$ sweeps through a solid angle that subtends exactly half the flux $\Phi_2$ during an exchange, accumulating half the full Aharonov-Bohm phase.
:::


:::{admonition} "Charge around flux" vs "flux around charge"
:class: attention

When composite A encircles composite B, one can equivalently view this as charge $q_1$ moving around flux $\Phi_2$, or flux $\Phi_1$ moving around charge $q_2$. These are **two descriptions of the same topological interaction** (the same linking of worldlines), not two separate contributions to be summed. The braiding phase is $q_1\Phi_2/\hbar$, counted once.
:::

:::{admonition} Recovering Bosons and Fermions
:class: note

- **Bosons**: $q\Phi/(2\hbar) = 0 \pmod{2\pi}$, so $\theta_{\mathrm{ex}} = 0$ and $s \in \mathbb{Z}$ (integer spin).
- **Fermions**: $q\Phi/(2\hbar) = \pi \pmod{2\pi}$, so $\theta_{\mathrm{ex}} = \pi$ and $s \in \mathbb{Z} + \tfrac{1}{2}$ (half-integer spin).
- **Anyons**: any other value gives fractional statistics and fractional spin.
:::

:::{admonition} Discussion: Why Does Spin = Exchange?
:class: dropdown tip

The spin-statistics relation $\theta_s = \theta_{\mathrm{ex}}$ follows naturally in the charge-flux picture: both self-rotation and exchange involve a charge sweeping past a flux in a half-braid topology. But in 3D, the spin-statistics theorem is proven from relativistic quantum field theory (CPT symmetry). Is the charge-flux argument a "derivation" of spin-statistics, or merely a consistency check? Can you think of a topological reason why self-rotation and exchange must give the same phase in 2D?
:::

:::{admonition} Derivation: Spin-Statistics Relation for Charge-Flux Composites
:class: dropdown information

We show that a $2\pi$ self-rotation of a charge-flux composite $(q, \Phi)$ produces the same phase as an exchange with an identical composite, establishing $\theta_s = \theta_{\mathrm{ex}}$.

**Self-rotation:** Rotate the composite by $2\pi$ about its center. The charge $q$ is displaced from the flux $\Phi$ by an infinitesimal distance (the composite has internal structure). During the $2\pi$ rotation, the charge traces a small loop that encloses the flux exactly once in a half-winding sense. More precisely, in the frame co-rotating with the composite, the charge sweeps an angle of $2\pi$ around the flux tube.

However, the key geometric identity is that a $2\pi$ self-rotation of a point-like composite is **topologically equivalent to an exchange** of two identical composites. To see this, consider two composites and imagine the following continuous deformation:

1. Start with two composites at $\boldsymbol{r}_1$ and $\boldsymbol{r}_2$.
2. Exchange them via a half-braid (each moves along a semicircle).
3. The relative coordinate $\boldsymbol{r} = \boldsymbol{r}_1 - \boldsymbol{r}_2$ rotates by $\pi$.
4. Now shrink composite 2 onto composite 1 — the half-braid becomes a $\pi$ self-rotation.

Since a half-braid = exchange = $\pi$ rotation of the relative coordinate, and the phase is $\theta_{\mathrm{ex}} = q\Phi/(2\hbar)$, we identify the spin as

$$
s = \frac{\theta_{\mathrm{ex}}}{2\pi} = \frac{q\Phi}{4\pi\hbar}
$$

and the $2\pi$ rotation phase (topological spin) is $\theta_s = 2\pi s = q\Phi/(2\hbar) = \theta_{\mathrm{ex}}$, confirming the spin-statistics relation.
:::

### Summary

| Dimension | Exchange group | Allowed statistics | Exchange squared |
|-----------|---------------|-------------------|------------------|
| 3D | Permutation $S_N$ | Bosons ($\theta=0$) or fermions ($\theta=\pi$) | $\hat{\mathcal{P}}^2 = I$ |
| 2D | Braid $B_N$ | Any $\theta \in [0, 2\pi)$ | $\hat{\mathcal{P}}^2 \neq I$ in general |

:::{admonition} Discussion: Can Anyons Exist in 3D?
:class: dropdown tip

In 3D, the topology of configuration space forces $\hat{\mathcal{P}}^2 = I$, seemingly ruling out anyons. But what about particles confined to a 2D layer embedded in 3D (like electrons in a quantum well)? Are these "truly 2D" for the purposes of exchange statistics? What physical conditions must be met for anyonic behavior to emerge?
:::
