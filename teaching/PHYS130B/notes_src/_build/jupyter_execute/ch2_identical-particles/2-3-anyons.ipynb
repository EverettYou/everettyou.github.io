# 2.3 Anyons

### Overview

| Dimension | Configuration Space | Allowed Statistics | Examples |
|-----------|--------------------|--------------------|----------|
| **3D+** | Simply connected | Boson, Fermion only | Electrons, photons |
| **2D** | Not simply connected | Anyons (continuous family) | FQHE quasiparticles |
| **1D** | Pathological | Tomonaga-Luttinger | TLL states |

**Abelian anyons**:

$$\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) \to e^{i\pi\theta} \psi(\boldsymbol{r}_2, \boldsymbol{r}_1)$$

where $\theta \in [0, 2)$ is the **statistical angle**.

- Bosons: $\theta = 0$ (phase $+1$)
- Fermions: $\theta = 1$ (phase $-1$)
- Fractional anyons: $\theta = 1/3$ (phase $e^{i\pi/3}$)

**Non-abelian anyons**: Braiding **mixes states** (not just phases). Higher dimensional R-matrices.

For $N$ identical anyons:

$$\dim(\mathcal{H}_N) \sim d^N$$

- Bosons/fermions: $d = 1$ (extensive Hilbert space, but symmetric subspace is 1-dimensional)
- Non-abelian Fibonacci anyons: $d = \phi = \frac{1+\sqrt{5}}{2} \approx 1.618$ (golden ratio!)

This gives **exponential advantage** in quantum information storage.
