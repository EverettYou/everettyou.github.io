## Lecture Notes

### Overview

When two systems each carry angular momentum — whether orbital and spin, or spin and spin — the total angular momentum $\hat{\boldsymbol{J}} = \hat{\boldsymbol{J}}_1 + \hat{\boldsymbol{J}}_2$ obeys addition rules with no classical analog. The combined Hilbert space admits two natural bases: the **uncoupled basis** $\vert j_1, m_1; j_2, m_2\rangle$ (individual projections known) and the **coupled basis** $\vert J, M\rangle$ (total angular momentum known). The **Clebsch-Gordan coefficients** provide the unitary change of basis, and the **triangle rule** determines which total angular momenta are allowed.

### Two Bases for the Same Space

The combined system lives in $\mathcal{H}_{j_1} \otimes \mathcal{H}_{j_2}$, with dimension $(2j_1+1)(2j_2+1)$.

**Uncoupled basis** $\vert j_1, m_1\rangle\vert j_2, m_2\rangle$: eigenstates of $\hat{J}_{1z}$ and $\hat{J}_{2z}$ separately. Good quantum numbers: $m_1, m_2$.

**Coupled basis** $\vert J, M\rangle$: eigenstates of $\hat{J}^2_{\text{tot}}$ and $\hat{J}_{z,\text{tot}}$. Good quantum numbers: $J, M$.

Both span the same space and are related by a unitary transformation.

:::{admonition} Triangle Rule
:class: important

When adding angular momenta $j_1$ and $j_2$, the allowed total angular momentum is:

$$
J \in \{\vert j_1 - j_2\vert,\; \vert j_1 - j_2\vert + 1,\; \ldots,\; j_1 + j_2\}
$$ (eq-triangle-rule)

Dimension check: $\displaystyle\sum_{J=\vert j_1-j_2\vert}^{j_1+j_2}(2J+1) = (2j_1+1)(2j_2+1)$.
:::

:::{admonition} Derivation: Triangle Rule
:class: dropdown information

**Step 1: Maximum $J$.** The highest possible $M = m_1 + m_2$ is $j_1 + j_2$ (when both projections are maximal). Only one uncoupled state achieves this: $\vert j_1, j_1\rangle\vert j_2, j_2\rangle$. This must be $\vert J_{\max}, J_{\max}\rangle$ with $J_{\max} = j_1 + j_2$.

**Step 2: Counting by subtraction.** At $M = j_1 + j_2 - 1$, there are two uncoupled states (one from lowering each $m$), but one belongs to the $J = j_1 + j_2$ multiplet. The remaining state starts a new multiplet with $J = j_1 + j_2 - 1$. Continuing: at each step, the number of uncoupled states with a given $M$ exceeds those already assigned, until we reach $M = \vert j_1 - j_2\vert$, after which no new multiplets appear.

**Step 3: Minimum $J$.** The process terminates at $J_{\min} = \vert j_1 - j_2\vert$, where the state count is exactly exhausted.

**Step 4: Dimension check.** The total number of states is:

$$
\sum_{J=\vert j_1 - j_2\vert}^{j_1+j_2}(2J+1)
$$

Setting $k = J - \vert j_1-j_2\vert$ and summing the arithmetic series gives $(2j_1+1)(2j_2+1)$, matching the dimension of $\mathcal{H}_{j_1}\otimes\mathcal{H}_{j_2}$. $\checkmark$
:::

### Clebsch-Gordan Coefficients

:::{admonition} Clebsch-Gordan Coefficients
:class: important

The CG coefficients $\langle j_1, m_1; j_2, m_2 \vert J, M\rangle$ express the coupled basis in terms of the uncoupled basis:

$$
\vert J, M\rangle = \sum_{m_1, m_2} \langle j_1, m_1; j_2, m_2 \vert J, M\rangle\;\vert j_1, m_1\rangle\vert j_2, m_2\rangle
$$ (eq-cg-expansion)

**Properties:**

- **Selection rule**: nonzero only if $M = m_1 + m_2$ and the triangle rule is satisfied
- **Reality**: all CG coefficients are real
- **Orthonormality**: $\sum_{m_1,m_2} \langle j_1, m_1; j_2, m_2 \vert J, M\rangle\langle j_1, m_1; j_2, m_2 \vert J', M'\rangle = \delta_{JJ'}\delta_{MM'}$
:::

In practice, CG coefficients are looked up in tables or computed by algebra systems. For simple cases they can be derived by applying ladder operators starting from the stretched state $\vert J_{\max}, J_{\max}\rangle = \vert j_1, j_1\rangle\vert j_2, j_2\rangle$.

:::{admonition} Derivation: Recursion Relations for Clebsch-Gordan Coefficients
:class: dropdown information

The CG coefficients can be computed systematically via recursion. The idea: apply $\hat{J}_-$ to the expansion of $\vert J,M\rangle$ and project onto uncoupled states.

**Notation.** Write $C(m_1, m_2\,|\,M) \equiv \langle j_1,m_1;j_2,m_2\vert J,M\rangle$ (suppressing fixed $j_1, j_2, J$).

**Step 1: Apply the total lowering operator.**

Start from $\vert J, M\rangle = \sum_{m_1+m_2=M} C(m_1,m_2\,|\,M)\,\vert j_1,m_1;j_2,m_2\rangle$ and apply $\hat{J}_- = \hat{J}_{1-} + \hat{J}_{2-}$:

$$
\hat{J}_-\vert J,M\rangle = \hbar\sqrt{J(J+1)-M(M-1)}\,\vert J,M-1\rangle
$$

On the right side, $\hat{J}_{1-}$ acts on particle 1 and $\hat{J}_{2-}$ on particle 2:

$$\begin{split}
&(\hat{J}_{1-}+\hat{J}_{2-})\vert j_1,m_1;j_2,m_2\rangle \\
=& \hbar\sqrt{j_1(j_1+1)-m_1(m_1-1)}\,\vert j_1,m_1{-}1;j_2,m_2\rangle \\
&+ \hbar\sqrt{j_2(j_2+1)-m_2(m_2-1)}\,\vert j_1,m_1;j_2,m_2{-}1\rangle
\end{split}
$$

**Step 2: Project onto $\langle j_1,m_1';j_2,m_2'\vert$ with $m_1'+m_2' = M-1$.**

Equating the $\vert j_1,m_1';j_2,m_2'\rangle$ component on both sides:

$$
\boxed{
\begin{split}    
&\sqrt{J(J+1)-M(M-1)}\;C(m_1',m_2'\,|\,M{-}1) \\
=& \sqrt{j_1(j_1+1)-m_1'(m_1'{+}1)}\;C(m_1'{+}1,m_2'\,|\,M) \\
&+ \sqrt{j_2(j_2+1)-m_2'(m_2'{+}1)}\;C(m_1',m_2'{+}1\,|\,M)
\end{split}}
$$

This is the **lowering recursion relation** — given the column $M$, it determines column $M-1$.

**Step 3: Starting condition.**

For each $J$, the unique state with $M = J$ and $m_1 = j_1$ (maximum) has:
$$
C(j_1,\,J-j_1\,|\,J) = 1 \quad \text{(positive by convention, normalized)}
$$

Applying the recursion downward (decreasing $M$) generates all $C(m_1', m_2'\,|\,M)$ for $M = J-1, J-2, \ldots, -J$.

**Step 4: Verification for $j_1 = j_2 = \tfrac{1}{2}$, $J = 1$.**

Start: $C(+\tfrac{1}{2},+\tfrac{1}{2}\,|\,1) = 1$. Apply recursion with $M=1, m_1' = -\tfrac{1}{2}, m_2' = +\tfrac{1}{2}$:

$$
\sqrt{1 \cdot 2 - 1 \cdot 0}\;C(-\tfrac{1}{2},+\tfrac{1}{2}\,|\,0) = \sqrt{\tfrac{3}{4}-(-\tfrac{1}{2})(+\tfrac{1}{2})}\; C(+\tfrac{1}{2},+\tfrac{1}{2}\,|\,1) + 0
$$

$$
\sqrt{2}\;C(-\tfrac{1}{2},+\tfrac{1}{2}\,|\,0) = 1 \implies C(-\tfrac{1}{2},+\tfrac{1}{2}\,|\,0) = \tfrac{1}{\sqrt{2}}
$$

By symmetry (or applying with $m_1' = +\tfrac{1}{2}, m_2' = -\tfrac{1}{2}$): $C(+\tfrac{1}{2},-\tfrac{1}{2}\,|\,0) = \tfrac{1}{\sqrt{2}}$. This gives $\vert 1,0\rangle = \tfrac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle + \vert\downarrow\uparrow\rangle)$, confirming the result in the key example. $\checkmark$
:::
### Key Example: Two Spin-1/2 Particles

For $j_1 = j_2 = 1/2$: $J \in \{0, 1\}$, giving $1 + 3 = 4 = 2 \times 2$ states.

:::{admonition} Triplet States ($J = 1$, Symmetric)
:class: important

$$
\vert 1, +1\rangle = \vert\uparrow\uparrow\rangle
$$

$$
\vert 1, 0\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle + \vert\downarrow\uparrow\rangle)
$$

$$
\vert 1, -1\rangle = \vert\downarrow\downarrow\rangle
$$ (eq-triplet)

All three are **symmetric** under particle exchange: $\hat{P}_{12}\vert 1, M\rangle = +\vert 1, M\rangle$.
:::

:::{admonition} Singlet State ($J = 0$, Antisymmetric)
:class: important

$$
\vert 0, 0\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle - \vert\downarrow\uparrow\rangle)
$$ (eq-singlet)

The singlet is **antisymmetric**: $\hat{P}_{12}\vert 0, 0\rangle = -\vert 0, 0\rangle$. It has **zero total spin** and is **maximally entangled** — measuring one particle's spin along any axis completely determines the other's.
:::

:::{admonition} Derivation: Two Spin-1/2 Clebsch-Gordan Coefficients
:class: dropdown information

**Stretched state.** The highest-weight state is unique: $\vert 1, 1\rangle = \vert\uparrow\uparrow\rangle$.

**Lowering to $M = 0$.** Apply $\hat{J}_- = \hat{J}_{1-} + \hat{J}_{2-}$ to $\vert 1, 1\rangle$:

$$
\hat{J}_-\vert 1,1\rangle = \hbar\sqrt{2}\,\vert 1, 0\rangle
$$

$$
(\hat{J}_{1-} + \hat{J}_{2-})\vert\uparrow\uparrow\rangle = \hbar\vert\downarrow\uparrow\rangle + \hbar\vert\uparrow\downarrow\rangle
$$

Equating: $\vert 1, 0\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle + \vert\downarrow\uparrow\rangle)$. Lowering again gives $\vert 1, -1\rangle = \vert\downarrow\downarrow\rangle$.

**Singlet by orthogonality.** At $M = 0$ there are two uncoupled states: $\vert\uparrow\downarrow\rangle$ and $\vert\downarrow\uparrow\rangle$. The $J = 0$ state must be orthogonal to $\vert 1, 0\rangle$ within this subspace:

$$
\vert 0, 0\rangle = \frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle - \vert\downarrow\uparrow\rangle)
$$

**Verification.** Check $\hat{J}_+\vert 0,0\rangle = 0$: $(\hat{J}_{1+} + \hat{J}_{2+})\frac{1}{\sqrt{2}}(\vert\uparrow\downarrow\rangle - \vert\downarrow\uparrow\rangle) = \frac{\hbar}{\sqrt{2}}(\vert\uparrow\uparrow\rangle - \vert\uparrow\uparrow\rangle) = 0$. $\checkmark$
:::

**Physical consequences:**

- **Helium ground state**: Two electrons in the $1s$ orbital have a symmetric spatial wavefunction, so the spin part must be the antisymmetric singlet — the electrons are spin-paired.
- **Pauli exclusion**: Identical fermions in the same spatial state must form a spin singlet.

:::{admonition} Discussion
:class: dropdown tip

The uncoupled state $\vert\uparrow\downarrow\rangle$ is an equal superposition of singlet and triplet ($M = 0$). If you measure $\hat{J}^2_{\text{tot}}$ on this state, you get $J = 0$ or $J = 1$ with equal probability. Before measuring, does the system "have" a definite total spin? What does this tell us about the relationship between individual and collective quantum numbers?
:::

### Application: Spin-Orbit Coupling

:::{admonition} Spin-Orbit Interaction
:class: important

An electron in an atom experiences a coupling between orbital and spin angular momentum:

$$
\hat{H}_{SO} = \lambda\,\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}
$$ (eq-spin-orbit)

Using $\hat{\boldsymbol{J}} = \hat{\boldsymbol{L}} + \hat{\boldsymbol{S}}$, we can write $\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}} = \frac{1}{2}(\hat{J}^2 - \hat{L}^2 - \hat{S}^2)$, which is **diagonal** in the coupled basis $\vert\ell, s; j, m_j\rangle$:

$$
\langle \hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}} \rangle = \frac{\hbar^2}{2}[j(j+1) - \ell(\ell+1) - s(s+1)]
$$
:::

:::{admonition} Derivation: $\hat{L}\cdot\hat{S}$ from $\hat{J}^2$
:class: dropdown information

From $\hat{\boldsymbol{J}} = \hat{\boldsymbol{L}} + \hat{\boldsymbol{S}}$, square both sides:

$$
\hat{J}^2 = (\hat{\boldsymbol{L}} + \hat{\boldsymbol{S}})^2 = \hat{L}^2 + \hat{S}^2 + 2\hat{\boldsymbol{L}}\cdot\hat{\boldsymbol{S}}
$$

where $\hat{\boldsymbol{L}}\cdot\hat{\boldsymbol{S}} = \hat{L}_x\hat{S}_x + \hat{L}_y\hat{S}_y + \hat{L}_z\hat{S}_z$ (the cross terms $\hat{L}_i\hat{S}_j$ with $i \neq j$ cancel in the expansion because $\hat{\boldsymbol{L}}$ and $\hat{\boldsymbol{S}}$ act on different Hilbert spaces and thus commute). Solving:

$$
\hat{\boldsymbol{L}}\cdot\hat{\boldsymbol{S}} = \frac{1}{2}(\hat{J}^2 - \hat{L}^2 - \hat{S}^2)
$$

In the coupled basis $\vert \ell, s; j, m_j\rangle$, all three operators on the right are diagonal, giving:

$$
\langle \hat{\boldsymbol{L}}\cdot\hat{\boldsymbol{S}}\rangle = \frac{\hbar^2}{2}[j(j+1) - \ell(\ell+1) - s(s+1)] \quad \checkmark
$$
:::

For an electron ($s = 1/2$) with orbital quantum number $\ell \geq 1$, two values are allowed: $j = \ell + 1/2$ and $j = \ell - 1/2$. Their energy splitting is the **fine structure** — a small but measurable correction. Example: the hydrogen $2p$ level splits into $2p_{3/2}$ and $2p_{1/2}$, a preview of perturbation theory in [§5.1](../ch5_perturbation-theory/5-1-time-independent-perturbation-theory).

### Summary

- **Two bases**: uncoupled ($m_1, m_2$ known) vs. coupled ($J, M$ known), related by CG coefficients.
- **Triangle rule**: $J \in \{\vert j_1 - j_2\vert, \ldots, j_1 + j_2\}$; dimension count: $\sum(2J+1) = (2j_1+1)(2j_2+1)$.
- **Clebsch-Gordan coefficients**: real, unitary, nonzero only when $M = m_1 + m_2$ and triangle rule holds.
- **Two spin-1/2**: triplet ($J = 1$, symmetric, 3 states) + singlet ($J = 0$, antisymmetric, maximally entangled).
- **Spin-orbit coupling**: $\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}$ is diagonal in the coupled basis; causes atomic fine structure.
