## Lecture Notes

Angular momentum is far more than a rotational property—it is a fundamental quantum number that governs the internal structure of atoms and the symmetry of space itself. Understanding its algebra reveals why particle spin and orbital motion are inextricably linked in the quantum world.
### Angular Momentum Operators

**Orbital angular momentum** is defined as:

$$

\boldsymbol{L} = \boldsymbol{r} \times \boldsymbol{p}

$$ (eq-orbital-angmom-def)

**Spin angular momentum** $\boldsymbol{S}$ is an intrinsic angular momentum with no classical analog.

**Total angular momentum**:

$$

\boldsymbol{J} = \boldsymbol{L} + \boldsymbol{S} \quad \text{(or more generally, any angular momentum)}

$$

Despite their different origins, all forms of angular momentum satisfy the **same fundamental commutation relations**.

### The Canonical Commutation Relations

The three components of any angular momentum operator satisfy:

$$

[J_i, J_j] = i\hbar \epsilon_{ijk} J_k

$$ (eq-angular-momentum-commutation)

where $\epsilon_{ijk}$ is the **Levi-Civita symbol** (completely antisymmetric):

$$

\epsilon_{123} = \epsilon_{231} = \epsilon_{312} = +1, \quad \epsilon_{321} = \epsilon_{213} = \epsilon_{132} = -1, \quad \text{else} = 0

$$

### Explicit Component Relations

The commutation relations break down to:

$$

[J_x, J_y] = i\hbar J_z, \quad [J_y, J_z] = i\hbar J_x, \quad [J_z, J_x] = i\hbar J_y

$$ (eq-explicit-commutators)

These are **cyclic** in $(x, y, z)$.

:::{admonition} Derivation: Angular Momentum Commutation Relation [J_i, J_j] = iℏε_ijk J_k
:class: dropdown note

**From orbital angular momentum definition:**

Let $J_i = L_i = \epsilon_{ijk} r_j p_k$ (summation over repeated indices).

$$

[L_i, L_j] = [\epsilon_{imn} r_m p_n, \epsilon_{jpq} r_p p_q]

$$

Expanding the commutator:

$$

[L_i, L_j] = \epsilon_{imn} \epsilon_{jpq} [r_m p_n, r_p p_q]

$$

Using $[r_m, r_p] = 0$, $[p_n, p_q] = 0$, and $[r_m, p_q] = i\hbar \delta_{mq}$:

$$

[r_m p_n, r_p p_q] = r_m [p_n, r_p] p_q + [r_m, r_p] p_n p_q + r_m r_p [p_n, p_q] + [r_m, p_q] p_n r_p

$$

$$

= r_m (-i\hbar \delta_{np}) p_q + 0 + 0 + i\hbar \delta_{mq} p_n r_p = i\hbar (\delta_{mq} p_n r_p - \delta_{np} r_m p_q)

$$

Therefore:

$$

[L_i, L_j] = i\hbar \epsilon_{imn} \epsilon_{jpq} (\delta_{mq} p_n r_p - \delta_{np} r_m p_q)

$$

Using the identity $\epsilon_{imn}\epsilon_{jpq}\delta_{mq} = \epsilon_{inn}\epsilon_{jpn}$ and simplifying with $\epsilon_{ijk}\epsilon_{imn} = \delta_{jm}\delta_{kn} - \delta_{jn}\delta_{km}$:

$$

[L_i, L_j] = i\hbar \epsilon_{ijk} L_k

$$

**For spin:** The commutation relation is **postulated** based on rotational invariance, and all angular momenta are required to satisfy the same algebra.
:::

### Why These Relations?

For orbital angular momentum, these relations follow from the canonical position-momentum commutation relation $[x_i, p_j] = i\hbar \delta_{ij}$. For spin, they are postulated based on the requirement of **rotational invariance**. The commutation relations encode the fundamental structure of rotations in quantum mechanics.

:::{admonition} Physical meaning of commutation relations
:class: note
The commutation relations $[J_i, J_j] = i\hbar \epsilon_{ijk} J_k$ encode the geometry of rotations: the order of rotations matters. This is the quantum manifestation of the non-commutativity of 3D rotations.
:::

## The Casimir Operator

### Definition

The **Casimir operator** (or **total angular momentum squared**) is:

$$

J^2 = J_x^2 + J_y^2 + J_z^2

$$ (eq-casimir-operator)

### Commutation with All Components

A remarkable property is that $J^2$ commutes with all components of $\boldsymbol{J}$:

$$

[J^2, J_x] = 0, \quad [J^2, J_y] = 0, \quad [J^2, J_z] = 0

$$ (eq-casimir-commutation)

::{admonition} Derivation: Commutation of Casimir with J_z
:class: dropdown note

**Proof** (for $[J^2, J_z]$):

$$

[J^2, J_z] = [J_x^2, J_z] + [J_y^2, J_z] + [J_z^2, J_z]

$$

The last term is zero. For the first:

$$

[J_x^2, J_z] = J_x[J_x, J_z] + [J_x, J_z]J_x = J_x(-i\hbar J_y) + (-i\hbar J_y)J_x = -i\hbar(J_x J_y + J_y J_x)

$$

$$

[J_y^2, J_z] = J_y[J_y, J_z] + [J_y, J_z]J_y = J_y(i\hbar J_x) + (i\hbar J_x)J_y = i\hbar(J_y J_x + J_x J_y)

$$

These cancel: $[J^2, J_z] = 0$. ✓
:::

### Simultaneous Eigenstates

Because $[J^2, J_z] = 0$, we can find simultaneous eigenstates of both $J^2$ and $J_z$. We label them as:

$$

|j, m\rangle

$$

with eigenvalues:

$$

J^2 |j, m\rangle = \hbar^2 j(j+1) |j, m\rangle

$$

$$

J_z |j, m\rangle = \hbar m |j, m\rangle

$$ (eq-angular-momentum-eigenstates)

where $j$ is the **angular momentum quantum number** and $m$ is the **magnetic quantum number** (projection along $z$).

## Ladder Operators (Raising and Lowering)

### Definition

The **raising operator** $J_+$ and **lowering operator** $J_-$ are defined as:

$$

J_+ = J_x + i J_y, \quad J_- = J_x - i J_y

$$ (eq-ladder-operators-def)

**Inverse relations**:

$$

J_x = \frac{1}{2}(J_+ + J_-), \quad J_y = \frac{1}{2i}(J_+ - J_-)

$$

### Commutation Relations for Ladder Operators

$$

[J_z, J_+] = \hbar J_+, \quad [J_z, J_-] = -\hbar J_-

$$ (eq-ladder-commutation-z)

$$

[J_+, J_-] = 2\hbar J_z

$$ (eq-ladder-commutation-pm)

$$

[J^2, J_\pm] = 0

$$ (eq-casimir-ladder)

### Action on Eigenstates

The ladder operators act on $|j, m\rangle$ as:

$$

J_+ |j, m\rangle = \hbar\sqrt{j(j+1) - m(m+1)} |j, m+1\rangle

$$ (eq-raising-action)

$$

J_- |j, m\rangle = \hbar\sqrt{j(j+1) - m(m-1)} |j, m-1\rangle

$$ (eq-lowering-action)

**Physical interpretation**:
- $J_+$ **raises** the $z$-projection of angular momentum: $m \to m+1$.
- $J_-$ **lowers** the $z$-projection: $m \to m-1$.
- The coefficient $\sqrt{\cdots}$ is a **normalization factor** ensuring that $|j, m+1\rangle$ is normalized if $|j, m\rangle$ is.

:::{admonition} Derivation: Action of Raising Operator J_+ on Eigenstates
:class: dropdown note

We prove: $J_+ |j, m\rangle = \hbar\sqrt{j(j+1) - m(m+1)} |j, m+1\rangle$

**Key relations:**
- $J_z |j, m\rangle = \hbar m |j, m\rangle$
- $J^2 |j, m\rangle = \hbar^2 j(j+1) |j, m\rangle$
- $[J_z, J_+] = \hbar J_+$
- $J^2$ commutes with $J_\pm$

**Argument:** Let $|\psi\rangle = J_+ |j, m\rangle$. We determine what state it is.

Apply $J_z$:

$$

J_z |\psi\rangle = J_z J_+ |j, m\rangle = (J_+ J_z + [J_z, J_+]) |j, m\rangle = J_+ J_z |j, m\rangle + \hbar J_+ |j, m\rangle

$$

$$

= J_+ (\hbar m) |j, m\rangle + \hbar J_+ |j, m\rangle = \hbar(m+1) J_+ |j, m\rangle = \hbar(m+1) |\psi\rangle

$$

So $|\psi\rangle$ is an eigenstate of $J_z$ with eigenvalue $\hbar(m+1)$, meaning it is proportional to $|j, m+1\rangle$.

Apply $J^2$:

$$

J^2 |\psi\rangle = J^2 J_+ |j, m\rangle = J_+ J^2 |j, m\rangle = J_+ (\hbar^2 j(j+1)) |j, m\rangle = \hbar^2 j(j+1) J_+ |j, m\rangle = \hbar^2 j(j+1) |\psi\rangle

$$

So $|\psi\rangle = |j, m+1\rangle$ (up to normalization).

**Normalization:** To find the coefficient, use the norm:

$$

\langle j, m | J_-^\dagger J_- |j, m\rangle = \langle j, m | J_+ J_- |j, m\rangle

$$

Using $J_+ J_- = J^2 - J_z^2 + \hbar J_z$:

$$

\langle j, m | (J^2 - J_z^2 + \hbar J_z) |j, m\rangle = \hbar^2 j(j+1) - \hbar^2 m^2 + \hbar^2 m = \hbar^2(j(j+1) - m^2 + m)

$$

$$

= \hbar^2(j(j+1) - m(m-1)) = \hbar^2((j-m)(j+m+1))

$$

Therefore:

$$

J_+ |j, m\rangle = \hbar\sqrt{(j-m)(j+m+1)} |j, m+1\rangle \quad ✓

$$

which can also be written as $\hbar\sqrt{j(j+1) - m(m+1)}$.
:::

### Normalization Coefficients

The square root coefficients can be rewritten using $j(j+1) - m(m+1) = (j-m)(j+m+1)$:

$$

J_+ |j, m\rangle = \hbar\sqrt{(j-m)(j+m+1)} |j, m+1\rangle

$$

Similarly, $j(j+1) - m(m-1) = (j+m)(j-m+1)$:

$$

J_- |j, m\rangle = \hbar\sqrt{(j+m)(j-m+1)} |j, m-1\rangle

$$

## Angular Momentum Quantization

### Derivation from Commutation Relations

The allowed values of angular momentum quantum numbers arise purely from the **algebra of commutation relations**, without solving any differential equations!

**Key observation**: For a state with angular momentum quantum number $j$, the eigenvalue equation $J_z |j, m\rangle = \hbar m |j, m\rangle$ requires that:

$$

m \in \{-j, -j+1, \ldots, j-1, j\}

$$

That is, $m$ ranges from $-j$ to $+j$ in integer steps. This has $2j+1$ possible values.

**For physical systems**, the requirement of **single-valuedness** (the wavefunction must return to itself after a $2\pi$ rotation) forces:

$$

j \in \{0, \frac{1}{2}, 1, \frac{3}{2}, 2, \ldots\}

$$

Both **integer** and **half-integer** values are allowed in quantum mechanics (unlike classical mechanics, where angular momentum is always an integer multiple of $\hbar$).

### Integer vs. Half-Integer

- **Integer $j$** ($j = 0, 1, 2, \ldots$): examples are orbital angular momentum, photons, pions.
- **Half-integer $j$** ($j = 1/2, 3/2, \ldots$): examples are electron spin, quarks, neutrinos.

The distinction has profound consequences through the **spin-statistics theorem**: integer-$j$ particles are bosons; half-integer-$j$ particles are fermions.

## Matrix Representations

### Spin-1/2 (Two-Dimensional Representation)

For $j = 1/2$, there are $2j + 1 = 2$ basis states: $|1/2, +1/2\rangle$ and $|1/2, -1/2\rangle$.

The angular momentum operators are represented as:

$$

J_z = \frac{\hbar}{2} \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} = \frac{\hbar}{2} \hat{\sigma}_z

$$

$$

J_+ = \hbar \begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}

$$

$$

J_- = \hbar \begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix}

$$

$$

J_x = \frac{\hbar}{2} \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} = \frac{\hbar}{2} \hat{\sigma}_x, \quad
J_y = \frac{\hbar}{2} \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix} = \frac{\hbar}{2} \hat{\sigma}_y

$$

where $\hat{\sigma}_i$ are the **Pauli matrices**.

### General Pattern

For a representation with angular momentum quantum number $j$, the operators are $(2j+1) \times (2j+1)$ matrices in the basis $\{|j, -j\rangle, |j, -j+1\rangle, \ldots, |j, j\rangle\}$.

All such representations are **irreducible** (cannot be block-diagonalized while respecting the angular momentum algebra).

## Orbital Angular Momentum (Specific Case)

### In Quantum Mechanics

Orbital angular momentum $\boldsymbol{L} = \boldsymbol{r} \times \boldsymbol{p}$ with components:

$$

L_x = y p_z - z p_y, \quad L_y = z p_x - x p_z, \quad L_z = x p_y - y p_x

$$ (eq-orbital-components)

In the position representation:

$$

L_z = -i\hbar \frac{\partial}{\partial \phi} \quad \text{(in spherical coordinates)}

$$

### Spherical Harmonics as Eigenfunctions

The eigenfunctions of $L^2$ and $L_z$ are the **spherical harmonics** $Y_\ell^m(\theta, \phi)$:

$$

L^2 Y_\ell^m = \hbar^2 \ell(\ell+1) Y_\ell^m

$$

$$

L_z Y_\ell^m = \hbar m Y_\ell^m

$$

where $\ell \in \{0, 1, 2, \ldots\}$ (always **integer**) and $m \in \{-\ell, \ldots, \ell\}$.

**Note**: Orbital angular momentum can only have integer quantum numbers. Half-integer is forbidden for $L$ (a consequence of requiring single-valuedness for the spatial wavefunction on a sphere).

## Useful Relations and Identities

### Useful Commutators

$$

[J_x, J_x] = 0 \quad \text{(self-commutation always zero)}

$$

For different components:

$$

[J_i, J_j] = i\hbar \epsilon_{ijk} J_k

$$

### Useful Products

$$

J_+ J_- = J^2 - J_z^2 + \hbar J_z

$$

$$

J_- J_+ = J^2 - J_z^2 - \hbar J_z

$$

::{admonition} Derivation: Verification of Ladder Product Relations
:class: dropdown note

These can be verified by expanding $J_\pm = J_x \pm i J_y$ and using $[J_x, J_y] = i\hbar J_z$.

Starting from $J_\pm = J_x \pm i J_y$:

$$

J_+ J_- = (J_x + i J_y)(J_x - i J_y) = J_x^2 - i J_x J_y + i J_y J_x + J_y^2

$$

$$

= J_x^2 + J_y^2 + i(J_y J_x - J_x J_y) = J_x^2 + J_y^2 - i[J_x, J_y]

$$

$$

= J_x^2 + J_y^2 - i \cdot i\hbar J_z = J_x^2 + J_y^2 + \hbar J_z

$$

Since $J^2 = J_x^2 + J_y^2 + J_z^2$, we have $J_x^2 + J_y^2 = J^2 - J_z^2$, giving:

$$

J_+ J_- = J^2 - J_z^2 + \hbar J_z

$$

The second relation follows similarly.
:::

### Completeness of Ladder Operators

Every state $|j, m\rangle$ can be reached from $|j, j\rangle$ (the state with maximum $m$) by repeatedly applying $J_-$:

$$

|j, m\rangle \propto (J_-)^{j-m} |j, j\rangle

$$

Similarly, every state can be reached from $|j, -j\rangle$ by applying $J_+$ repeatedly.

### Expectation Values

For a state $|j, m\rangle$:

$$

\langle J_x \rangle = \langle J_y \rangle = 0 \quad \text{(no preferred $x$ or $y$ direction)}

$$

$$

\langle J_z \rangle = \hbar m

$$

$$

\langle J^2 \rangle = \hbar^2 j(j+1)

$$

## Summary

- **Angular momentum algebra**: All angular momenta satisfy $[J_i, J_j] = i\hbar \epsilon_{ijk} J_k$.
- **Casimir operator** $J^2 = J_x^2 + J_y^2 + J_z^2$ commutes with all components; simultaneous eigenstates exist: $|j, m\rangle$.
- **Eigenvalues**: $J^2 |j, m\rangle = \hbar^2 j(j+1) |j, m\rangle$ and $J_z |j, m\rangle = \hbar m |j, m\rangle$.
- **Quantization**: $j \in \{0, 1/2, 1, 3/2, \ldots\}$ and $m \in \{-j, -j+1, \ldots, j-1, j\}$ (integer steps).
- **Ladder operators** $J_\pm = J_x \pm i J_y$ shift $m$ by one unit: $J_\pm |j, m\rangle = \hbar\sqrt{(j\mp m)(j\pm m+1)} |j, m\pm 1\rangle$.
- **Orbital angular momentum** always has integer $\ell$; **spin** can be half-integer.
- **Representation matrices**: For spin-1/2, use Pauli matrices; general $j$ requires $(2j+1) \times (2j+1)$ matrices.
