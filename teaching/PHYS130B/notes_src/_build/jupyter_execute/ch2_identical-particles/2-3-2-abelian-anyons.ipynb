## Lecture Notes

:::{admonition} Key Concept: Abelian Anyons
:class: important

In two-dimensional quantum systems, the fundamental symmetry group changes, allowing particles to acquire fractional charge and exchange phase factors more general than $\pm 1$. These **abelian anyons** underlie the fractional quantum Hall effect, one of the most profound discoveries in condensed matter physics.
:::

### Definition

Abelian anyons are the simplest type of anyonic excitations. When two abelian anyons are exchanged, the wavefunction is multiplied by a **phase factor** (a complex number of modulus 1):

$$


\psi(\boldsymbol{r}_1, \boldsymbol{r}_2) \to \mathrm{e}^{\mathrm{i}\theta} \psi(\boldsymbol{r}_2, \boldsymbol{r}_1)

$$ (eq-abelian-exchange)

The word "abelian" refers to the fact that exchanging particles in different orders gives the same result up to overall phases—the operations **commute**.

**Key difference from bosons and fermions:**
- Bosons: $\theta = 0$ (no phase upon exchange)
- Fermions: $\theta = \pi$ (phase of $-1$ upon exchange)
- Anyons: $0 < \theta < 2\pi$ (fractional phases)

### Contrast with Non-Abelian Anyons

Non-abelian anyons, by contrast, can **mix different quantum states** under braiding. Exchanging two non-abelian anyons in different orders leads to **different states** (not just phase differences), making them far more complex but also more powerful for quantum computation.

### Why Study Abelian Anyons?

- **Simpler mathematics**: Phase factors are easier to handle than matrix transformations
- **Experimentally realized**: Abelian anyons have been observed in quantum Hall systems
- **Foundation for non-abelian**: Understanding abelian cases builds intuition for richer non-abelian structure

:::{admonition} Discussion: Measuring Fractional Statistics
:class: dropdown warning

The phase $\mathrm{e}^{\mathrm{i}\theta}$ picked up by exchanging two anyons is measurable only through **two-anyon interferometry**—braiding paths that cross versus paths that don't cross.

**Question**: Without interference, is $\theta$ directly observable, or is it an artifact of how we compute amplitudes? If two anyonic excitations have $\theta = \pi/2$ (between fermionic and bosonic), can we tell from the energy spectrum, lifetime, or other single-particle properties that they carry fractional statistics? What observable distinguishes $\theta = \pi$ (fermion) from $\theta = \pi/2$ (anyon) without looping one around the other?
:::

### Key Properties

- **Fractional charge**: Quasiparticles in quantum Hall states carry fractional charge (e.g., $e/3$ in the $\nu = 1/3$ state)
- **Fractional angular momentum**: Related to braiding statistics
- **Braiding: only difference from bosons/fermions**: Abelian anyons satisfy braiding relations but commute when separated

### Summary

- In **two spatial dimensions**, particle exchange can produce a phase factor $\mathrm{e}^{\mathrm{i}\theta}$ instead of $\pm 1$; these are **anyons**
- **Abelian anyons** accumulate phase $\theta$ for each exchange; for $N$ identical anyons, statistics phase depends only on the number of exchanges
- **Fractional statistics emerge**: Bosons ($\theta = 0$), fermions ($\theta = \pi$), intermediate anyons with $0 < \theta < 2\pi$
- **Anyonic excitations** can have fractional charge and fractional angular momentum
- **Examples**: Laughlin quasiholes in fractional quantum Hall effect at filling factor $\nu = 1/m$ have charge $e/m$ and exchange phase $\mathrm{e}^{\mathrm{i}2\pi/m}$
- **2D phenomenon**: Braiding statistics is fundamentally 2D; in 3D, only $\pm 1$ exchange phases survive
## The Aharonov-Bohm Effect

### Electromagnetic Potentials and Gauge Invariance

In classical electromagnetism, the **electric and magnetic fields** $\boldsymbol{E}, \boldsymbol{B}$ are the physical observables; the potentials $\phi, \boldsymbol{A}$ are auxiliary tools.

In quantum mechanics, the **potentials** become fundamental: the electron wavefunction couples directly to the electromagnetic potential via:

$$


\boldsymbol{p} \to \boldsymbol{p} - \frac{e}{c}\boldsymbol{A} \quad \text{(Minimal coupling)}

$$


### The Aharonov-Bohm Phenomenon

Consider a solenoid (magnetic field confined to a thin cylindrical region). Outside the solenoid, $\boldsymbol{B} = 0$, but the vector potential $\boldsymbol{A} \neq 0$.

An electron traveling around the solenoid in regions where $\boldsymbol{B} = 0$ should (classically) be unaffected. Yet in quantum mechanics, it acquires a **phase shift**:

$$


\Delta \phi_{AB} = -\frac{e}{\hbar c} \oint \boldsymbol{A} \cdot d\boldsymbol{\ell} = -\frac{e}{\hbar c} \Phi_{\text{enclosed}}

$$ (eq-aharonov-bohm-phase)

where $\Phi_{\text{enclosed}}$ is the magnetic flux through the solenoid.

**Physical consequence**: The electron's wavefunction phase is altered by a region where the electron never travels!

### Connection to Anyons

Anyonic statistics can be understood as a generalization of the Aharonov-Bohm effect:

- Imagine each anyon carries a **microscopic flux tube** with quantized magnetic flux.
- When two anyons exchange, each passes through the flux tube of the other.
- The wavefunction phase picked up is proportional to the flux.

This is the **flux-charge composite** picture (see below).
## Flux-Charge Composites

### The Idea

A simple way to construct abelian anyons is to combine **charges** and **fluxes**.

**Flux tube**: A thin (ideally infinitesimal) solenoid carrying a quantized magnetic flux $\Phi$.

**Composite**: Attach a charged particle (charge $q$) to a flux tube carrying flux $\Phi$.

### Exchange Phase

When two flux-charge composites exchange:

1. The first composite (charge $q_1$, flux $\Phi_1$) passes through the flux tube of the second composite (flux $\Phi_2$).
2. The Aharonov-Bohm phase acquired is: $\Delta \phi_1 = -\frac{q_1}{\hbar c} \Phi_2$.
3. By symmetry, the second composite acquires phase: $\Delta \phi_2 = -\frac{q_2}{\hbar c} \Phi_1$.
4. **Total exchange phase**:

$$


\phi_{\text{exch}} = \frac{q_1 \Phi_2 + q_2 \Phi_1}{\hbar c} = \frac{2\pi}{\Phi_0} (q_1 \Phi_2 + q_2 \Phi_1)

$$


where $\Phi_0 = h/e$ is the **flux quantum** (in units where $c = 1$).

::{admonition} Derivation: Fractional Exchange Phase from Flux-Charge Composites
:class: dropdown note

**Example**: Consider electrons ($q = -e$) each bound to a quantized flux $\Phi = m \Phi_0$ (where $m$ is an integer).

The exchange phase is:

$$


\mathrm{e}^{\mathrm{i}\phi} = \exp\left[i \frac{2\pi}{\Phi_0} \cdot (-e) \cdot m\Phi_0\right] = \mathrm{e}^{-2\pi i m}

$$


Wait—this gives $\mathrm{e}^{-2\pi i m} = 1$ (no phase), which seems wrong. The resolution is that **flux quantization** and **fractional filling** lead to fractional fluxes being attached.

**Better example** ($\nu = 1/3$ filling): An electron can be associated with $\Phi = \Phi_0/3$ (one-third of a flux quantum). Then:

$$


\phi = 2\pi \cdot \frac{1}{3} = \frac{2\pi}{3}

$$


The exchange phase is $\mathrm{e}^{\mathrm{i} 2\pi/3}$, corresponding to **statistical angle** $\theta = 2/3$.
:::
## The Fractional Quantum Hall Effect (FQHE)

### Setup: Quantum Hall Effect

A 2D electron gas in a strong perpendicular magnetic field exhibits the **quantum Hall effect**: the Hall conductance is quantized:

$$


\hat{\sigma}_{xy} = \nu \frac{e^2}{h}

$$ (eq-hall-conductance)

where $\nu$ is the **filling factor** (number of filled Landau levels or fraction thereof).

### Integer vs. Fractional

**Integer quantum Hall effect** ($\nu$ integer): Explained by non-interacting electrons filling discrete Landau levels. Robust but relatively simple.

**Fractional quantum Hall effect** ($\nu = p/q$ with small integers $p, q$, $\gcd(p,q)=1$): Arises from **electron-electron interactions** and collective quantum fluid behavior. Anomalous and fascinating.

### Filling Factors and Quasiparticles

At filling factor $\nu = 1/m$ (odd integer $m$), there is a special **gapped ground state** discovered by Laughlin.

**Excitations** (quasiparticles) in this state have:
- **Fractional charge**: $q_{qp} = e/m$ (one $m$-th of an electron charge).
- **Fractional statistics**: Exchange phase $\mathrm{e}^{\mathrm{i}\pi\theta}$ with $\theta = 2/m$ (for the Laughlin state at $\nu = 1/m$).

These quasiparticles are **abelian anyons**.
## The Laughlin Wavefunction

### Construction

For $N$ electrons in a strong magnetic field at filling factor $\nu = 1/m$, Laughlin proposed the ground state wavefunction:

$$


\Psi(z_1, z_2, \ldots, z_N) = \prod_{i<j}(z_i - z_j)^m \exp\left(-\sum_{k=1}^N \frac{|z_k|^2}{4\ell_B^2}\right)

$$ (eq-laughlin-wavefunction)

where:
- $z_j = x_j + iy_j$ are complex coordinates for the positions in 2D.
- The product $\prod_{i<j}(z_i - z_j)^m$ enforces correlations: electrons avoid each other (zero if two electrons are at the same point when $m > 0$).
- The exponential is a **Gaussian envelope** (ground state of a 2D harmonic oscillator) with length scale $\ell_B = \sqrt{\hbar/(eB)}$ (magnetic length).

### Properties

:::{admonition} Key Properties of the Laughlin State
:class: important

1. **Incompressible**: The ground state cannot be compressed—adding or removing electrons costs energy (creates gapped quasiparticles).

2. **Topologically ordered**: Long-range entanglement; the state cannot be described by a product or sum of local degrees of freedom.

3. **No single-particle description**: The Laughlin state cannot be expressed as a Slater determinant of orbitals (it is many-body entangled).

4. **Robust**: Protected by a finite energy gap; small perturbations don't destroy it.
:::

### Ground State Energy

The Laughlin wavefunction is believed to be the exact ground state (or very close) at $\nu = 1/m$ for certain interactions. It was a breakthrough in understanding the FQHE.
## Quasiparticles in the FQHE

### Quasiholes and Quasielectrons

**Quasihole**: Created by removing an electron from the ground state (or leaving a "missing electron" at position $z_h$). In the Laughlin wavefunction, a quasihole at position $z_h$ is created by multiplying the ground state by:

$$


\prod_{k=1}^N (z_h - z_k)^{-m}

$$


**Quasielectron**: Created by adding an extra electron. In the Laughlin state picture, it is represented by an additional factor in the wavefunction.

### Charge of Quasiparticles

:::{admonition} Fractional Quasiparticle Charges
:class: important

At $\nu = 1/m$:
- **Quasihole charge**: $+e/m$ (removing an electron leaves a positive charge of $1/m$ electron charge).
- **Quasielectron charge**: $-e/m$ (effectively an "extra electron" shared among $m$ positions).
:::

### Statistical Angle

:::{admonition} Quasihole Braiding Statistics at $\nu = 1/m$
:class: important

When two quasiholes exchange, the wavefunction acquires phase:

$$


\mathrm{e}^{\mathrm{i} \pi \theta} \quad \text{with} \quad \theta = \frac{2}{m}

$$


For example, at $\nu = 1/3$: $\theta = 2/3$, so $\mathrm{e}^{\mathrm{i} 2\pi/3}$ (a phase of $120°$).
:::

### Composite Fermion Picture

An alternative (and very useful) perspective is the **composite fermion** picture:
- Each electron is viewed as composite: a "bare" electron plus attached flux quanta.
- At $\nu = 1/m$, each electron binds $m$ flux quanta.
- The composite fermions fill an integer number of Landau levels (hence the integer quantum Hall effect appears for composite fermions).

This picture explains why the FQHE exhibits certain regularities and provides intuition for excitations.
:::{admonition} Example: Fractional Filling Factors
:class: example


### $\nu = 1/3$ (Most Famous)

- **Ground state**: Laughlin $\nu = 1/3$ state.
- **Quasihole charge**: $+e/3$.
- **Quasihole statistics**: $\theta = 2/3$.
- **Experimental observation**: Clear quantized Hall conductance at $\hat{\sigma}_{xy} = (1/3) e^2/h$.

### $\nu = 1/5, 1/7, \ldots$ (Odd Denominator)

All odd-denominator fractions $\nu = 1/m$ are stable and show FQHE.

### $\nu = 2/3, 3/5, \ldots$ (Composite Fractions)

These correspond to holes in the $\nu = 1/3, 2/5, \ldots$ states and are also experimentally observed.

### Even Denominators ($\nu = 1/2, 1/4, \ldots$)

These are more subtle and were not initially understood. Some (like $\nu = 5/2$) are now understood as non-abelian states (see Section 2.3.3).
:::
## Topological Order and Robustness

### What Is Topological Order?

**Topological order** is long-range quantum entanglement that is:
- **Robust to local perturbations**: The ground state is protected by an energy gap; disorder doesn't destabilize it.
- **Non-local**: Information about the system is stored in global (topological) properties, not in local correlations.
- **Characterized by anyons**: The anyonic excitations are the signature of topological order.

### Ground State Degeneracy

On a torus (or other non-simply connected surfaces), the FQHE ground state is **degenerate**: there are multiple orthogonal ground states.

The degeneracy is related to the number of types of anyons, providing another signature of topological order.

### Edge States

At the **boundary** of a FQHE sample, there are **chiral edge states**: electrons that can only move in one direction (say, counterclockwise around the boundary).

These edge states are **gapless** (no energy gap for excitations) and carry the Hall current. They are protected topologically by the bulk gap and are extremely robust against disorder.
## Experimental Realization

### Quantum Hall Effect in 2D Electron Gases

The fractional quantum Hall effect is most clearly observed in:
- **2D electron gases** (2DEG) in semiconductor heterostructures (e.g., GaAs/AlGaAs).
- **Graphene** (single and bilayer).
- **Topological insulators** (surface and edge states).

### Measurement Signatures

1. **Conductance quantization**: Measured in transport experiments; Hall conductance plateaus appear at rational filling factors.

2. **Energy gaps**: Spectroscopic measurements show gapped excitation spectrum.

3. **Anyonic signatures**: Fractional charges and statistics are harder to measure directly but can be inferred from noise and shot noise measurements.

### Challenges

- Extremely clean samples required (disorder destroys the effect).
- Ultra-low temperatures (millikelvin range) needed.
- Strong magnetic fields (several Tesla).

Despite these challenges, FQHE is well-established and among the most interesting quantum many-body phenomena.
## Summary

- **Abelian anyons**: Exchange phase $\mathrm{e}^{\mathrm{i}\phi}$ with arbitrary $\phi$; simplest non-bosonic/fermionic statistics.
- **Aharonov-Bohm effect**: Electrons gain phase passing through flux tubes; underlies anyon statistics.
- **Flux-charge composites**: Binding charge to flux tubes naturally creates anyonic statistics.
- **Fractional quantum Hall effect**: 2D electrons in strong $B$-field exhibit FQHE at special filling factors $\nu = p/q$.
- **Laughlin wavefunction**: Ground state at $\nu = 1/m$ with fractional charge and statistics quasiparticles.
- **Quasiholes**: Charge $+e/m$, statistics $\theta = 2/m$ at $\nu = 1/m$.
- **Topological order**: FQHE ground state is protected by gap; anyons signature of topological structure.
- **Experimental**: Well-observed in 2DEG, graphene, topological insulators; among most interesting quantum phenomena.
