## Lecture Notes

### Overview

Particles of the same type (electrons, photons, etc.) are absolutely identical in quantum mechanics—they carry no identity tags and are completely interchangeable. This indistinguishability is not a practical limitation but a fundamental principle that restricts which quantum states are physical: multi-particle states must be either symmetric (bosons) or antisymmetric (fermions) under particle exchange.

**Analogy.** Money in a bank account is identical. You cannot ask "Is this the dollar I deposited today or yesterday?"—dollars are indistinguishable. Identical particles are the same: two electrons have no labels. This simple observation has profound consequences for the structure of matter.


### Identical Particles

:::{admonition} Identical Particles
:class: important

Particles of the same type have identical mass, charge, spin, and **no inherent labels**. This indistinguishability is a fundamental principle that constrains which quantum states are physical.
:::

**Physical requirement.** Let $\hat{\mathcal{P}}_{ij}$ be the permutation operator that swaps particles $i$ and $j$. For identical particles, all observables must be invariant under permutation: $\langle\Psi\vert \hat{O}\vert\Psi\rangle = \langle\Psi\vert \hat{\mathcal{P}}_{ij}^\dagger \hat{O}\,\hat{\mathcal{P}}_{ij}\vert\Psi\rangle$ for any observable $\hat{O}$. This requires the state to be an eigenstate of every transposition:

$$

\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = \lambda\,\vert\Psi\rangle

$$

Since $\hat{\mathcal{P}}_{ij}^2 = \hat{I}$, the eigenvalue satisfies $\lambda^2 = 1$, so $\lambda = \pm 1$.

:::{admonition} Symmetry Constraint
:class: important

Under particle exchange, the many-body state must satisfy:

- **Symmetric** ($\lambda = +1$, **bosons**): $\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = +\vert\Psi\rangle$
- **Antisymmetric** ($\lambda = -1$, **fermions**): $\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = -\vert\Psi\rangle$

The choice is determined by particle spin (integer → boson, half-integer → fermion) via the spin-statistics theorem.
:::

:::{admonition} Derivation: Exchange Eigenvalues Are $\pm 1$
:class: dropdown information

**Step 1: Involution property.** The permutation operator $\hat{\mathcal{P}}_{ij}$ swaps particles $i$ and $j$. Swapping twice returns to the original state, so $\hat{\mathcal{P}}_{ij}^2 = \hat{I}$.

**Step 2: Eigenvalue constraint.** Let $\vert\Psi\rangle$ be an eigenstate with eigenvalue $\lambda$: $\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = \lambda\vert\Psi\rangle$. Applying $\hat{\mathcal{P}}_{ij}$ again:

$$

\hat{\mathcal{P}}_{ij}^2\vert\Psi\rangle = \lambda^2\vert\Psi\rangle = \vert\Psi\rangle

$$

so $\lambda^2 = 1$, giving $\lambda = +1$ (symmetric) or $\lambda = -1$ (antisymmetric).

**Step 3: Why must the state be an eigenstate?** For identical particles, every observable $\hat{O}$ commutes with $\hat{\mathcal{P}}_{ij}$: $[\hat{O}, \hat{\mathcal{P}}_{ij}] = 0$. This means the Hamiltonian preserves the symmetric and antisymmetric subspaces separately, so a system that starts in one subspace stays there forever. Nature chooses one subspace for each particle species.

**Step 4: Unitarity.** Since $\hat{\mathcal{P}}_{ij}^2 = \hat{I}$ and $\hat{\mathcal{P}}_{ij}$ preserves inner products (it just relabels particles), $\hat{\mathcal{P}}_{ij}$ is both unitary and Hermitian: $\hat{\mathcal{P}}_{ij}^\dagger = \hat{\mathcal{P}}_{ij} = \hat{\mathcal{P}}_{ij}^{-1}$.
:::

:::{admonition} Discussion: Indistinguishability and Entanglement
:class: dropdown tip

A symmetrized or antisymmetrized state is always an entangled superposition—you cannot assign a definite single-particle state to a specific particle. Is this "entanglement" the same as in a Bell pair, or is it fundamentally different? Can you design an experiment that distinguishes indistinguishability-entanglement from preparation-entanglement?
:::


### Symmetric and Antisymmetric States

Let $\{\vert\alpha\rangle\}$ be an orthonormal basis for the single-particle Hilbert space $\mathcal{H}$. A generic $N$-particle state in $\mathcal{H}^{\otimes N}$ is

$$

\vert\Psi\rangle = \sum_{\alpha_1, \ldots, \alpha_N} \Psi_{\alpha_1 \cdots \alpha_N}\;\vert\alpha_1\rangle \otimes \vert\alpha_2\rangle \otimes \cdots \otimes \vert\alpha_N\rangle

$$ (eq-many-body-state)

where $\Psi_{\alpha_1 \cdots \alpha_N}$ is the many-body amplitude. For identical particles, only the symmetric or antisymmetric subspace of $\mathcal{H}^{\otimes N}$ is physical.

#### Bosonic (Symmetric) States

For $N$ bosons placed in *distinct* single-particle states $\vert\alpha_1\rangle, \ldots, \vert\alpha_N\rangle$, the symmetrized state is:

$$

\vert\alpha_1, \ldots, \alpha_N\rangle_+ = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \vert\alpha_{\pi(1)}\rangle \otimes \vert\alpha_{\pi(2)}\rangle \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle

$$ (eq-boson-symmetrized)

The sum (all signs positive) is the ket-space analog of the **permanent**. Multiple bosons can occupy the same state.

:::{admonition} Example: Two Bosons
:class: dropdown example

**Problem.** Construct the symmetric state for two bosons in orthogonal states $\vert\alpha_1\rangle, \vert\alpha_2\rangle$.

**Solution.**

$$

\vert\alpha_1, \alpha_2\rangle_+ = \frac{1}{\sqrt{2}} \bigl(\vert\alpha_1\rangle\vert\alpha_2\rangle + \vert\alpha_2\rangle\vert\alpha_1\rangle\bigr)

$$

If both bosons occupy the same state $\vert\alpha\rangle$: $\vert\alpha, \alpha\rangle_+ = \vert\alpha\rangle\vert\alpha\rangle$ (already symmetric).
:::

:::{admonition} Normalization with Repeated States
:class: attention

More generally, when $n_\alpha$ bosons share the same single-particle state, the permanent acquires $n_\alpha !$ identical terms, which should be further normalized by $\sqrt{n_\alpha !}$. So for $N$ bosons with occupation numbers $\{n_\alpha\}$, the normalization factor is:

$$

\mathcal{N}_+ = \frac{1}{\sqrt{N! \prod_\alpha n_\alpha!}}

$$

:::

:::{admonition} Derivation: Normalization of Symmetrized States
:class: dropdown information

**Distinct states.** For $N$ bosons in mutually orthogonal states $\vert\alpha_1\rangle, \ldots, \vert\alpha_N\rangle$, the unnormalized symmetrized state is $\vert\tilde{\Psi}\rangle = \sum_{\pi \in S_N} \vert\alpha_{\pi(1)}\rangle \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle$. Its norm squared is:

$$

\langle\tilde{\Psi}\vert\tilde{\Psi}\rangle = \sum_{\pi, \pi'} \prod_{k=1}^N \langle\alpha_{\pi'(k)}\vert\alpha_{\pi(k)}\rangle = \sum_{\pi, \pi'} \prod_{k=1}^N \delta_{\pi'(k),\pi(k)}

$$

The product of Kronecker deltas equals 1 only when $\pi' = \pi$. Since there are $N!$ permutations, $\langle\tilde{\Psi}\vert\tilde{\Psi}\rangle = N!$, so the normalization factor is $1/\sqrt{N!}$.

**Repeated states.** When $n_\alpha$ bosons share state $\vert\alpha\rangle$, different permutations $\pi$ and $\pi'$ can produce the same tensor product state (permutations that only rearrange particles within the same mode are invisible). For each mode $\alpha$, there are $n_\alpha!$ such redundant permutations, so each distinct term appears $\prod_\alpha n_\alpha!$ times. The norm squared becomes $\langle\tilde{\Psi}\vert\tilde{\Psi}\rangle = N! \prod_\alpha n_\alpha!$, giving the normalization factor:

$$

\mathcal{N}_+ = \frac{1}{\sqrt{N! \prod_\alpha n_\alpha!}}

$$

**Fermionic case.** For antisymmetrized states with distinct single-particle states, the norm calculation is similar: $\langle\tilde{\Psi}\vert\tilde{\Psi}\rangle = \sum_{\pi,\pi'} \mathrm{sgn}(\pi)\,\mathrm{sgn}(\pi') \prod_k \delta_{\pi'(k),\pi(k)} = \sum_\pi [\mathrm{sgn}(\pi)]^2 = N!$. The normalization is again $1/\sqrt{N!}$. No repeated-state correction is needed because fermionic states with $n_\alpha \geq 2$ vanish identically.
:::

#### Fermionic (Antisymmetric) States

For $N$ fermions placed in *distinct* single-particle states $\vert\alpha_1\rangle, \ldots, \vert\alpha_N\rangle$, the antisymmetrized state is:

$$

\vert\alpha_1, \ldots, \alpha_N\rangle_- = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \mathrm{sgn}(\pi)\;\vert\alpha_{\pi(1)}\rangle \otimes \vert\alpha_{\pi(2)}\rangle \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle

$$ (eq-fermion-antisymmetrized)

This is the ket-space analog of the **Slater determinant**. Antisymmetry is automatic: swapping any two particle labels flips the sign. If $\vert\alpha_i\rangle = \vert\alpha_j\rangle$ for $i \neq j$, the state vanishes—this is the Pauli exclusion principle.

:::{admonition} Derivation: Slater Determinant Representation
:class: dropdown information

The antisymmetrized state can be written as a determinant. Define the $N \times N$ matrix $M$ with entries $M_{ij} = \vert\alpha_j\rangle_i$ (particle $i$ in state $\alpha_j$). The determinant formula gives:

$$

\det M = \sum_{\pi \in S_N} \mathrm{sgn}(\pi)\;\vert\alpha_{\pi(1)}\rangle_1 \otimes \vert\alpha_{\pi(2)}\rangle_2 \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle_N

$$

which is exactly the unnormalized antisymmetrized state. So:

$$

\vert\alpha_1, \ldots, \alpha_N\rangle_- = \frac{1}{\sqrt{N!}} \begin{vmatrix} \vert\alpha_1\rangle_1 & \vert\alpha_2\rangle_1 & \cdots & \vert\alpha_N\rangle_1 \\ \vert\alpha_1\rangle_2 & \vert\alpha_2\rangle_2 & \cdots & \vert\alpha_N\rangle_2 \\ \vdots & \vdots & \ddots & \vdots \\ \vert\alpha_1\rangle_N & \vert\alpha_2\rangle_N & \cdots & \vert\alpha_N\rangle_N \end{vmatrix}

$$

**Why antisymmetry is automatic:** Swapping two particles (rows) flips the sign of a determinant. Setting two states equal (columns) makes the determinant vanish—this is Pauli exclusion as a property of linear algebra.

For wavefunctions, the Slater determinant takes the familiar form:

$$

\Psi(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_N) = \frac{1}{\sqrt{N!}} \begin{vmatrix} \psi_1(\boldsymbol{r}_1) & \psi_2(\boldsymbol{r}_1) & \cdots & \psi_N(\boldsymbol{r}_1) \\ \psi_1(\boldsymbol{r}_2) & \psi_2(\boldsymbol{r}_2) & \cdots & \psi_N(\boldsymbol{r}_2) \\ \vdots & \vdots & \ddots & \vdots \\ \psi_1(\boldsymbol{r}_N) & \psi_2(\boldsymbol{r}_N) & \cdots & \psi_N(\boldsymbol{r}_N) \end{vmatrix}

$$

where $\psi_j(\boldsymbol{r}_i) = \langle\boldsymbol{r}_i\vert\alpha_j\rangle$ are single-particle wavefunctions.
:::

:::{admonition} Example: Two Fermions
:class: dropdown example

**Problem.** Construct the antisymmetric state for two fermions in states $\vert\alpha_1\rangle, \vert\alpha_2\rangle$.

**Solution.**

$$

\vert\alpha_1, \alpha_2\rangle_- = \frac{1}{\sqrt{2}} \bigl(\vert\alpha_1\rangle\vert\alpha_2\rangle - \vert\alpha_2\rangle\vert\alpha_1\rangle\bigr)

$$

**Check**: Swap particle labels gives $-\vert\alpha_1, \alpha_2\rangle_-$. Set $\vert\alpha_1\rangle = \vert\alpha_2\rangle$ gives zero.
:::


### Pauli Exclusion Principle

:::{admonition} Pauli Exclusion Principle
:class: important

No two identical **fermions** can occupy the same single-particle quantum state. The fermionic occupation number satisfies $n_\alpha \in \{0, 1\}$.

(**Bosons** have no such restriction: $n_\alpha \in \{0, 1, 2, \ldots\}$)
:::

This follows directly from antisymmetry: if $\vert\alpha_i\rangle = \vert\alpha_j\rangle$, the antisymmetrized state has two identical columns in the Slater determinant and vanishes. The consequences are far-reaching: atomic shell structure, chemical bonding, degeneracy pressure in white dwarfs and neutron stars, and the stability of matter itself.

:::{admonition} Derivation: Pauli Exclusion Principle from Antisymmetry
:class: dropdown information

We prove that no two fermions can occupy the same single-particle state by showing that the antisymmetrized state vanishes when any two modes coincide.

**Setup.** Consider $N$ fermions with states $\vert\alpha_1, \alpha_2, \ldots, \alpha_N\rangle_{-}$, where the antisymmetrized $N$-particle state is:

$$
\vert\alpha_1, \alpha_2, \ldots, \alpha_N\rangle_{-} = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \text{sgn}(\pi) \, \vert\alpha_{\pi(1)}\rangle \otimes \vert\alpha_{\pi(2)}\rangle \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle
$$

**Step 1: Apply the exchange operator when two states coincide.**

Suppose particles $i$ and $j$ are in the same state: $\alpha_i = \alpha_j \equiv \alpha$. Apply the exchange operator $\hat{\mathcal{P}}_{ij}$ (which swaps particles $i$ and $j$):

$$
\hat{\mathcal{P}}_{ij} \vert\alpha_1, \ldots, \alpha_i, \ldots, \alpha_j, \ldots, \alpha_N\rangle_{-} = \vert\alpha_1, \ldots, \alpha_j, \ldots, \alpha_i, \ldots, \alpha_N\rangle_{-}
$$

But since $\alpha_i = \alpha_j = \alpha$, the right-hand side is **identical to the original state**:

$$
\hat{\mathcal{P}}_{ij} \vert\ldots\rangle_{-} = \vert\ldots\rangle_{-}
$$

**Step 2: Use antisymmetry of fermionic states.**

Fermionic states are antisymmetric under any particle exchange, so:

$$
\hat{\mathcal{P}}_{ij} \vert\ldots\rangle_{-} = -\vert\ldots\rangle_{-}
$$

**Step 3: Combine.**

From Steps 1 and 2:

$$
\vert\ldots\rangle_{-} = -\vert\ldots\rangle_{-} \implies 2\vert\ldots\rangle_{-} = 0 \implies \vert\ldots\rangle_{-} = 0
$$

The state vanishes identically.

**Slater determinant perspective.** In the Slater determinant representation, if $\alpha_i = \alpha_j$, two rows of the matrix are identical ($\vert\alpha\rangle$ appears twice). A matrix with two identical rows has zero determinant, so $\vert\alpha_1, \ldots, \alpha_N\rangle_{-} = 0$ directly.

**Occupation number perspective.** In second quantization, applying the fermionic creation operator twice: $\hat{c}^\dagger_\alpha \hat{c}^\dagger_\alpha = 0$ (from $\{\hat{c}_\alpha, \hat{c}^\dagger_\alpha\} = 1 \Rightarrow \hat{c}^\dagger_\alpha \hat{c}^\dagger_\alpha = -\hat{c}^\dagger_\alpha \hat{c}^\dagger_\alpha \Rightarrow \hat{c}^\dagger_\alpha \hat{c}^\dagger_\alpha = 0$). Hence occupation number $n_\alpha \in \{0, 1\}$ only.
:::



:::{admonition} Derivation: Dimensions of the Symmetric and Antisymmetric Subspaces
:class: dropdown information

How many independent $N$-particle states are there when the particles are identical? The answer follows from counting distinct occupation configurations.

**Setup.** Let there be $d$ single-particle modes (orthonormal states) $\vert\alpha_1\rangle, \ldots, \vert\alpha_d\rangle$. Distribute $N$ identical particles among them.

**Fermionic case** (antisymmetric subspace):

Antisymmetry requires all occupation numbers $n_\alpha \in \{0,1\}$: no two fermions can share a state (Pauli exclusion). A basis state is fully determined by choosing which $N$ of the $d$ modes are occupied — an unordered selection of $N$ from $d$ without repetition. The number of such choices is:

$$
\dim\mathcal{H}_{\text{antisym}}^{(N)} = \binom{d}{N} = \frac{d!}{N!(d-N)!}
$$

This is the number of linearly independent Slater determinants. For $N > d$, there are no valid configurations: $\binom{d}{N} = 0$.

**Bosonic case** (symmetric subspace):

Occupation numbers $n_\alpha \in \{0, 1, 2, \ldots\}$ with $\sum_\alpha n_\alpha = N$. A basis state is an unordered selection of $N$ from $d$ modes **with repetition** (bosons can pile into the same mode). The number of such distributions is given by the "stars and bars" formula:

$$
\dim\mathcal{H}_{\text{sym}}^{(N)} = \binom{N+d-1}{N} = \frac{(N+d-1)!}{N!\,(d-1)!}
$$

*Proof by stars and bars:* Represent $N$ particles as stars $\star$ and use $d-1$ bars $\vert$ to divide them among $d$ modes. For example, with $d=3$ modes and $N=4$ particles: $\star\star\vert\star\vert\star$ means $(n_1,n_2,n_3)=(2,1,1)$. The total number of arrangements of $N$ stars and $d-1$ bars among $N+d-1$ positions is $\binom{N+d-1}{N}$. $\square$

**Consistency check:** For one particle ($N=1$), both subspaces must have dimension $d$ (a single particle cannot be "symmetrized"):
- Fermionic: $\binom{d}{1} = d$ ✓
- Bosonic: $\binom{d}{1} = d$ ✓

**Two-particle qubit example** ($d=2$, $N=2$):
- Bosonic: $\binom{3}{2}=3$ states: $\vert\alpha_1\alpha_1\rangle$, $\tfrac{1}{\sqrt{2}}(\vert\alpha_1\alpha_2\rangle+\vert\alpha_2\alpha_1\rangle)$, $\vert\alpha_2\alpha_2\rangle$ (the spin triplet).
- Fermionic: $\binom{2}{2}=1$ state: $\tfrac{1}{\sqrt{2}}(\vert\alpha_1\alpha_2\rangle-\vert\alpha_2\alpha_1\rangle)$ (the spin singlet).
- Check: $3 + 1 = 4 = d^N = 2^2$ — the symmetric and antisymmetric subspaces together span the full tensor product. $\checkmark$

**Large-$N$ asymptotics:**
- Fermionic: $\binom{d}{N} \leq \binom{d}{d/2} \sim 2^d/\sqrt{d}$ — bounded even as $N\to d$.
- Bosonic: $\binom{N+d-1}{N} \sim N^{d-1}/(d-1)!$ for fixed $d$, $N\to\infty$ — grows polynomially.

These count how much of the "exponentially large" $d^N$-dimensional tensor product space is physically accessible for indistinguishable particles.
:::

### Insertion and Deletion Rules

To add or remove a particle while preserving symmetry, we define **insertion** ($\rhd_\pm$) and **deletion** ($\lhd_\pm$) operations on first-quantized states. Let $\vert\alpha\rangle$ be a single-particle state, let $\mathbb{1}$ be the tensor identity (generator of the zero-particle space $\mathbb{C}$, satisfying $\vert\alpha\rangle \equiv \mathbb{1} \otimes \vert\alpha\rangle \equiv \vert\alpha\rangle \otimes \mathbb{1}$), and let $\vert\Psi\rangle = \vert\alpha_1\rangle \otimes \vert\alpha_2\rangle \otimes \cdots$ be a generic tensor product state. The subscript $\pm$ denotes symmetrization (+, bosons) or antisymmetrization (−, fermions).

:::{admonition} Insertion and Deletion Rules
:class: important

**Insertion** $\rhd_\pm$ (adds a particle symmetrically/antisymmetrically):

$$

\vert\alpha\rangle \rhd_\pm \mathbb{1} = \vert\alpha\rangle, \qquad \vert\alpha\rangle \rhd_\pm (\vert\beta\rangle \otimes \vert\Psi\rangle) = \vert\alpha\rangle \otimes \vert\beta\rangle \otimes \vert\Psi\rangle \pm \vert\beta\rangle \otimes (\vert\alpha\rangle \rhd_\pm \vert\Psi\rangle)

$$

**Deletion** $\lhd_\pm$ (removes a particle symmetrically/antisymmetrically):

$$

\vert\alpha\rangle \lhd_\pm \mathbb{1} = 0, \qquad \vert\alpha\rangle \lhd_\pm (\vert\beta\rangle \otimes \vert\Psi\rangle) = \delta_{\alpha\beta}\,\vert\Psi\rangle \pm \vert\beta\rangle \otimes (\vert\alpha\rangle \lhd_\pm \vert\Psi\rangle)

$$

Both operations are **linear** in the state being acted on.
:::

The recursive structure means insertion places $\vert\alpha\rangle$ at every possible position with alternating signs (for fermions) or all-positive signs (for bosons). Deletion removes $\vert\alpha\rangle$ from every position where it appears, with the same sign pattern.

:::{admonition} Example: Insertion into a Two-Particle State
:class: dropdown example

**Problem.** Compute $\vert\alpha\rangle \rhd_\pm (\vert\beta\rangle \otimes \vert\gamma\rangle)$.

**Solution.** Apply the recursive relation twice:

$$

\vert\alpha\rangle \rhd_\pm (\vert\beta\rangle \otimes \vert\gamma\rangle) = \vert\alpha\rangle\vert\beta\rangle\vert\gamma\rangle \pm \vert\beta\rangle \otimes (\vert\alpha\rangle \rhd_\pm \vert\gamma\rangle)

$$

For the inner insertion, use $\vert\gamma\rangle = \vert\gamma\rangle \otimes \mathbb{1}$:

$$

\vert\alpha\rangle \rhd_\pm \vert\gamma\rangle = \vert\alpha\rangle\vert\gamma\rangle \pm \vert\gamma\rangle\vert\alpha\rangle

$$

So:

$$

\vert\alpha\rangle \rhd_\pm (\vert\beta\rangle\vert\gamma\rangle) = \vert\alpha\rangle\vert\beta\rangle\vert\gamma\rangle \pm \vert\beta\rangle\vert\alpha\rangle\vert\gamma\rangle \pm \vert\beta\rangle\vert\gamma\rangle\vert\alpha\rangle

$$

(dropping $\otimes$ for brevity). For bosons (+): sum of all 3 cyclic insertions. For fermions (−): alternating signs $+1, -1, +1$.
:::

The insertion and deletion rules $\rhd_\pm$ and $\lhd_\pm$ are the first-quantized building blocks for adding and removing particles. When properly normalized, they yield **creation** ($\hat{a}^\dagger_\alpha$) and **annihilation** ($\hat{a}_\alpha$) operators—the central objects of second quantization, developed in [§2.1.3 Second Quantization](2-1-3-second-quantization).

### Summary

- **Identical particles** have no labels; the many-body state must be symmetric (bosons, $\lambda=+1$) or antisymmetric (fermions, $\lambda=-1$) under particle exchange.
- **Symmetric states** (permanent): sum over all permutations with positive signs; bosons allow unlimited occupation.
- **Antisymmetric states** (Slater determinant): alternating-sign sum; the Pauli exclusion principle ($n_\alpha \in \{0,1\}$) follows automatically.
- **Insertion/deletion rules** $\rhd_\pm$, $\lhd_\pm$ recursively add/remove particles while preserving symmetry. When normalized, they become the creation and annihilation operators of second quantization (§2.1.3).
