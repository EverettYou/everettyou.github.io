## Lecture Notes

### Overview

Particles of the same type (electrons, photons, etc.) are absolutely identical in quantum mechanics—they carry no identity tags and are completely interchangeable. This indistinguishability is not a practical limitation but a fundamental principle that restricts which quantum states are physical: multi-particle states must be either symmetric (bosons) or antisymmetric (fermions) under particle exchange.

**Analogy.** Money in a bank account is identical. You cannot ask "Is this the dollar I deposited today or yesterday?"—dollars are indistinguishable. Identical particles are the same: two electrons have no labels. This simple observation has profound consequences for the structure of matter.


### Identical Particles

:::{admonition} Identical Particles
:class: important

Particles of the same type have identical mass, charge, spin, and **no inherent labels**. This indistinguishability is a fundamental principle that constrains which quantum states are physical.
:::

**Physical requirement.** Let $\hat{\mathcal{P}}_{ij}$ be the permutation operator that swaps particles $i$ and $j$. For identical particles, all observables must be invariant under permutation: $\langle\Psi\vert \hat{O}\vert\Psi\rangle = \langle\Psi\vert \hat{\mathcal{P}}_{ij}^\dagger \hat{O}\,\hat{\mathcal{P}}_{ij}\vert\Psi\rangle$ for any observable $\hat{O}$. This requires the state to be an eigenstate of every transposition:

$$

\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = \lambda\,\vert\Psi\rangle

$$

Since $\hat{\mathcal{P}}_{ij}^2 = \hat{I}$, the eigenvalue satisfies $\lambda^2 = 1$, so $\lambda = \pm 1$.

:::{admonition} Symmetry Constraint
:class: important

Under particle exchange, the many-body state must satisfy:

- **Symmetric** ($\lambda = +1$, **bosons**): $\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = +\vert\Psi\rangle$
- **Antisymmetric** ($\lambda = -1$, **fermions**): $\hat{\mathcal{P}}_{ij}\vert\Psi\rangle = -\vert\Psi\rangle$

The choice is determined by particle spin (integer → boson, half-integer → fermion) via the spin-statistics theorem.
:::

:::{admonition} Discussion: Indistinguishability and Entanglement
:class: dropdown tip

A symmetrized or antisymmetrized state is always an entangled superposition—you cannot assign a definite single-particle state to a specific particle. Is this "entanglement" the same as in a Bell pair, or is it fundamentally different? Can you design an experiment that distinguishes indistinguishability-entanglement from preparation-entanglement?
:::


### Symmetric and Antisymmetric States

Let $\{\vert\alpha\rangle\}$ be an orthonormal basis for the single-particle Hilbert space $\mathcal{H}$. A generic $N$-particle state in $\mathcal{H}^{\otimes N}$ is

$$

\vert\Psi\rangle = \sum_{\alpha_1, \ldots, \alpha_N} \Psi_{\alpha_1 \cdots \alpha_N}\;\vert\alpha_1\rangle \otimes \vert\alpha_2\rangle \otimes \cdots \otimes \vert\alpha_N\rangle

$$ (eq-many-body-state)

where $\Psi_{\alpha_1 \cdots \alpha_N}$ is the many-body amplitude. For identical particles, only the symmetric or antisymmetric subspace of $\mathcal{H}^{\otimes N}$ is physical.

#### Bosonic (Symmetric) States

For $N$ bosons placed in *distinct* single-particle states $\vert\alpha_1\rangle, \ldots, \vert\alpha_N\rangle$, the symmetrized state is:

$$

\vert\alpha_1, \ldots, \alpha_N\rangle_+ = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \vert\alpha_{\pi(1)}\rangle \otimes \vert\alpha_{\pi(2)}\rangle \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle

$$ (eq-boson-symmetrized)

The sum (all signs positive) is the ket-space analog of the **permanent**. Multiple bosons can occupy the same state.

:::{admonition} Example: Two Bosons
:class: dropdown example

**Problem.** Construct the symmetric state for two bosons in orthogonal states $\vert\alpha_1\rangle, \vert\alpha_2\rangle$.

**Solution.**

$$

\vert\alpha_1, \alpha_2\rangle_+ = \frac{1}{\sqrt{2}} \bigl(\vert\alpha_1\rangle\vert\alpha_2\rangle + \vert\alpha_2\rangle\vert\alpha_1\rangle\bigr)

$$

If both bosons occupy the same state $\vert\alpha\rangle$: $\vert\alpha, \alpha\rangle_+ = \vert\alpha\rangle\vert\alpha\rangle$ (already symmetric).
:::

:::{admonition} Normalization with Repeated States
:class: attention

More generally, when $n_\alpha$ bosons share the same single-particle state, the permanent acquires $n_\alpha !$ identical terms, which should be further normalized by $\sqrt{n_\alpha !}$. So for $N$ bosons with occupation numbers $\{n_\alpha\}$, the normalization factor is:

$$

\mathcal{N}_+ = \frac{1}{\sqrt{N! \prod_\alpha n_\alpha!}}

$$

:::

#### Fermionic (Antisymmetric) States

For $N$ fermions placed in *distinct* single-particle states $\vert\alpha_1\rangle, \ldots, \vert\alpha_N\rangle$, the antisymmetrized state is:

$$

\vert\alpha_1, \ldots, \alpha_N\rangle_- = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \mathrm{sgn}(\pi)\;\vert\alpha_{\pi(1)}\rangle \otimes \vert\alpha_{\pi(2)}\rangle \otimes \cdots \otimes \vert\alpha_{\pi(N)}\rangle

$$ (eq-fermion-antisymmetrized)

This is the ket-space analog of the **Slater determinant**. Antisymmetry is automatic: swapping any two particle labels flips the sign. If $\vert\alpha_i\rangle = \vert\alpha_j\rangle$ for $i \neq j$, the state vanishes—this is the Pauli exclusion principle.

:::{admonition} Example: Two Fermions
:class: dropdown example

**Problem.** Construct the antisymmetric state for two fermions in states $\vert\alpha_1\rangle, \vert\alpha_2\rangle$.

**Solution.**

$$

\vert\alpha_1, \alpha_2\rangle_- = \frac{1}{\sqrt{2}} \bigl(\vert\alpha_1\rangle\vert\alpha_2\rangle - \vert\alpha_2\rangle\vert\alpha_1\rangle\bigr)

$$

**Check**: Swap particle labels gives $-\vert\alpha_1, \alpha_2\rangle_-$. Set $\vert\alpha_1\rangle = \vert\alpha_2\rangle$ gives zero.
:::


### Pauli Exclusion Principle

:::{admonition} Pauli Exclusion Principle
:class: important

No two identical **fermions** can occupy the same single-particle quantum state. The fermionic occupation number satisfies $n_\alpha \in \{0, 1\}$.

(**Bosons** have no such restriction: $n_\alpha \in \{0, 1, 2, \ldots\}$)
:::

This follows directly from antisymmetry: if $\vert\alpha_i\rangle = \vert\alpha_j\rangle$, the antisymmetrized state has two identical columns in the Slater determinant and vanishes. The consequences are far-reaching: atomic shell structure, chemical bonding, degeneracy pressure in white dwarfs and neutron stars, and the stability of matter itself.


### Insertion and Deletion Rules

To add or remove a particle while preserving symmetry, we define **insertion** ($\rhd_\pm$) and **deletion** ($\lhd_\pm$) operations on first-quantized states. Let $\vert\alpha\rangle$ be a single-particle state, let $\mathbb{1}$ be the tensor identity (generator of the zero-particle space $\mathbb{C}$, satisfying $\vert\alpha\rangle \equiv \mathbb{1} \otimes \vert\alpha\rangle \equiv \vert\alpha\rangle \otimes \mathbb{1}$), and let $\vert\Psi\rangle = \vert\alpha_1\rangle \otimes \vert\alpha_2\rangle \otimes \cdots$ be a generic tensor product state. The subscript $\pm$ denotes symmetrization (+, bosons) or antisymmetrization (−, fermions).

:::{admonition} Insertion and Deletion Rules
:class: important

**Insertion** $\rhd_\pm$ (adds a particle symmetrically/antisymmetrically):

$$

\vert\alpha\rangle \rhd_\pm \mathbb{1} = \vert\alpha\rangle, \qquad \vert\alpha\rangle \rhd_\pm (\vert\beta\rangle \otimes \vert\Psi\rangle) = \vert\alpha\rangle \otimes \vert\beta\rangle \otimes \vert\Psi\rangle \pm \vert\beta\rangle \otimes (\vert\alpha\rangle \rhd_\pm \vert\Psi\rangle)

$$

**Deletion** $\lhd_\pm$ (removes a particle symmetrically/antisymmetrically):

$$

\vert\alpha\rangle \lhd_\pm \mathbb{1} = 0, \qquad \vert\alpha\rangle \lhd_\pm (\vert\beta\rangle \otimes \vert\Psi\rangle) = \delta_{\alpha\beta}\,\vert\Psi\rangle \pm \vert\beta\rangle \otimes (\vert\alpha\rangle \lhd_\pm \vert\Psi\rangle)

$$

Both operations are **linear** in the state being acted on.
:::

The recursive structure means insertion places $\vert\alpha\rangle$ at every possible position with alternating signs (for fermions) or all-positive signs (for bosons). Deletion removes $\vert\alpha\rangle$ from every position where it appears, with the same sign pattern.

:::{admonition} Example: Insertion into a Two-Particle State
:class: dropdown example

**Problem.** Compute $\vert\alpha\rangle \rhd_\pm (\vert\beta\rangle \otimes \vert\gamma\rangle)$.

**Solution.** Apply the recursive relation twice:

$$

\vert\alpha\rangle \rhd_\pm (\vert\beta\rangle \otimes \vert\gamma\rangle) = \vert\alpha\rangle\vert\beta\rangle\vert\gamma\rangle \pm \vert\beta\rangle \otimes (\vert\alpha\rangle \rhd_\pm \vert\gamma\rangle)

$$

For the inner insertion, use $\vert\gamma\rangle = \vert\gamma\rangle \otimes \mathbb{1}$:

$$

\vert\alpha\rangle \rhd_\pm \vert\gamma\rangle = \vert\alpha\rangle\vert\gamma\rangle \pm \vert\gamma\rangle\vert\alpha\rangle

$$

So:

$$

\vert\alpha\rangle \rhd_\pm (\vert\beta\rangle\vert\gamma\rangle) = \vert\alpha\rangle\vert\beta\rangle\vert\gamma\rangle \pm \vert\beta\rangle\vert\alpha\rangle\vert\gamma\rangle \pm \vert\beta\rangle\vert\gamma\rangle\vert\alpha\rangle

$$

(dropping $\otimes$ for brevity). For bosons (+): sum of all 3 cyclic insertions. For fermions (−): alternating signs $+1, -1, +1$.
:::


#### Boson Creation and Annihilation Operators

The **boson creation operator** $\hat{b}^\dagger_\alpha$ and **annihilation operator** $\hat{b}_\alpha$ are defined by normalizing the insertion and deletion operations acting on an $N$-particle state $\vert\Psi\rangle$:

:::{admonition} Definition: Boson Creation and Annihilation
:class: important

$$

\hat{b}^\dagger_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N+1}}\;\vert\alpha\rangle \rhd_+ \vert\Psi\rangle, \qquad \hat{b}_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N}}\;\vert\alpha\rangle \lhd_+ \vert\Psi\rangle

$$ (eq-boson-op-def)

The factor $1/\sqrt{N+1}$ (resp. $1/\sqrt{N}$) normalizes the $(N{+}1)$-particle (resp. $(N{-}1)$-particle) state produced by symmetric insertion (resp. deletion).
:::

:::{admonition} Example: $\hat{b}^\dagger_1 \vert 1_1, 1_2\rangle = \sqrt{2}\,\vert 2_1, 1_2\rangle$
:class: dropdown example

**Problem.** Create one more boson in state $\vert 1\rangle$, starting from $\vert 1_1, 1_2\rangle = (\vert 1\rangle\vert 2\rangle + \vert 2\rangle\vert 1\rangle)/\sqrt{2}$.

**Solution.** Apply the definition $\hat{b}^\dagger_1 = \frac{1}{\sqrt{3}}\vert 1\rangle \rhd_+$ (since $N=2$, so $N+1=3$):

$$

\hat{b}^\dagger_1 \vert 1_1, 1_2\rangle = \frac{1}{\sqrt{2}} \left( \frac{1}{\sqrt{3}}\vert 1\rangle \rhd_+ \vert 1\rangle\vert 2\rangle + \frac{1}{\sqrt{3}}\vert 1\rangle \rhd_+ \vert 2\rangle\vert 1\rangle \right)

$$

Compute each insertion using the recursive rule:

$$

\vert 1\rangle \rhd_+ \vert 1\rangle\vert 2\rangle = \vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle(\vert 1\rangle \rhd_+ \vert 2\rangle) = \vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 2\rangle\vert 1\rangle

$$

$$

= 2\,\vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 2\rangle\vert 1\rangle

$$

$$

\vert 1\rangle \rhd_+ \vert 2\rangle\vert 1\rangle = \vert 1\rangle\vert 2\rangle\vert 1\rangle + \vert 2\rangle(\vert 1\rangle \rhd_+ \vert 1\rangle) = \vert 1\rangle\vert 2\rangle\vert 1\rangle + \vert 2\rangle\vert 1\rangle\vert 1\rangle + \vert 2\rangle\vert 1\rangle\vert 1\rangle

$$

$$

= \vert 1\rangle\vert 2\rangle\vert 1\rangle + 2\,\vert 2\rangle\vert 1\rangle\vert 1\rangle

$$

Sum:

$$

\hat{b}^\dagger_1 \vert 1_1, 1_2\rangle = \frac{2}{\sqrt{6}} (\vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 2\rangle\vert 1\rangle + \vert 2\rangle\vert 1\rangle\vert 1\rangle)

$$

Recognizing $\vert 2_1, 1_2\rangle = \frac{1}{\sqrt{3}} (\vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 2\rangle\vert 1\rangle + \vert 2\rangle\vert 1\rangle\vert 1\rangle)$:

$$

\hat{b}^\dagger_1 \vert 1_1, 1_2\rangle = \frac{2\sqrt{3}}{\sqrt{6}}\;\vert 2_1, 1_2\rangle = \sqrt{2}\;\vert 2_1, 1_2\rangle

$$

This confirms $\hat{b}^\dagger_1 \vert n_1, n_2\rangle = \sqrt{n_1+1}\;\vert n_1+1, n_2\rangle$ with $n_1 = 1$.
:::

:::{admonition} Example: $\hat{b}_1 \vert 2_1, 1_2\rangle = \sqrt{2}\,\vert 1_1, 1_2\rangle$
:class: dropdown example

**Problem.** Annihilate a boson from state $\vert 1\rangle$, starting from $\vert 2_1, 1_2\rangle = \frac{1}{\sqrt{3}}(\vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 2\rangle\vert 1\rangle + \vert 2\rangle\vert 1\rangle\vert 1\rangle)$.

**Solution.** Apply the definition $\hat{b}_1 = \frac{1}{\sqrt{3}}\vert 1\rangle \lhd_+$ (since $N=3$):

$$

\hat{b}_1 \vert 2_1, 1_2\rangle = \frac{1}{3}\bigl(\vert 1\rangle \lhd_+ \vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle \lhd_+ \vert 1\rangle\vert 2\rangle\vert 1\rangle + \vert 1\rangle \lhd_+ \vert 2\rangle\vert 1\rangle\vert 1\rangle\bigr)

$$

Compute each deletion using $\vert\alpha\rangle \lhd_+ (\vert\beta\rangle \otimes \vert\Psi\rangle) = \delta_{\alpha\beta}\vert\Psi\rangle + \vert\beta\rangle \otimes (\vert\alpha\rangle \lhd_+ \vert\Psi\rangle)$:

$$

\vert 1\rangle \lhd_+ \vert 1\rangle\vert 1\rangle\vert 2\rangle = \vert 1\rangle\vert 2\rangle + \vert 1\rangle(\vert 1\rangle \lhd_+ \vert 1\rangle\vert 2\rangle) = \vert 1\rangle\vert 2\rangle + \vert 1\rangle(\vert 2\rangle + 0) = 2\,\vert 1\rangle\vert 2\rangle

$$

$$

\vert 1\rangle \lhd_+ \vert 1\rangle\vert 2\rangle\vert 1\rangle = \vert 2\rangle\vert 1\rangle + \vert 1\rangle(\vert 1\rangle \lhd_+ \vert 2\rangle\vert 1\rangle) = \vert 2\rangle\vert 1\rangle + \vert 1\rangle(0 + \vert 2\rangle) = \vert 2\rangle\vert 1\rangle + \vert 1\rangle\vert 2\rangle

$$

$$

\vert 1\rangle \lhd_+ \vert 2\rangle\vert 1\rangle\vert 1\rangle = 0 + \vert 2\rangle(\vert 1\rangle \lhd_+ \vert 1\rangle\vert 1\rangle) = \vert 2\rangle(\vert 1\rangle + \vert 1\rangle) = 2\,\vert 2\rangle\vert 1\rangle

$$

Sum: $\frac{1}{3}(3\,\vert 1\rangle\vert 2\rangle + 3\,\vert 2\rangle\vert 1\rangle) = \vert 1\rangle\vert 2\rangle + \vert 2\rangle\vert 1\rangle = \sqrt{2}\,\vert 1_1, 1_2\rangle$.
:::

**Action on single-mode Fock states.** Starting from the vacuum $\vert 0_\alpha\rangle = \mathbb{1}$ and applying $\hat{b}^\dagger_\alpha$ repeatedly, one can derive the action on any Fock state:

:::{admonition} Boson Operators on Fock States
:class: important

$$

\hat{b}^\dagger_\alpha \vert n_\alpha\rangle = \sqrt{n_\alpha + 1}\;\vert n_\alpha + 1\rangle, \qquad \hat{b}_\alpha \vert n_\alpha\rangle = \sqrt{n_\alpha}\;\vert n_\alpha - 1\rangle, \qquad \hat{b}_\alpha \vert 0\rangle = 0

$$ (eq-boson-fock)

:::

:::{admonition} Derivation: Action on Single-Mode Fock States
:class: dropdown note

The single-mode Fock state with $n$ bosons is $\vert n_\alpha\rangle = \vert\alpha\rangle^{\otimes n} / \sqrt{n!}$ (after normalization of the permanent). From the definition:

$$

\hat{b}^\dagger_\alpha \vert n_\alpha\rangle = \frac{1}{\sqrt{n+1}} \;\vert\alpha\rangle \rhd_+ \frac{\vert\alpha\rangle^{\otimes n}}{\sqrt{n!}}

$$

Since all $n$ particles are in the same state $\vert\alpha\rangle$, the symmetric insertion $\vert\alpha\rangle \rhd_+ \vert\alpha\rangle^{\otimes n}$ produces $(n+1)$ identical copies of $\vert\alpha\rangle^{\otimes(n+1)}$ (one for each insertion position, all with $+$ sign). So:

$$

\vert\alpha\rangle \rhd_+ \vert\alpha\rangle^{\otimes n} = (n+1)\;\vert\alpha\rangle^{\otimes(n+1)}

$$

Therefore:

$$

\hat{b}^\dagger_\alpha \vert n_\alpha\rangle = \frac{(n+1)}{\sqrt{(n+1)\cdot n!}}\;\vert\alpha\rangle^{\otimes(n+1)} = \sqrt{n+1}\;\frac{\vert\alpha\rangle^{\otimes(n+1)}}{\sqrt{(n+1)!}} = \sqrt{n+1}\;\vert n+1\rangle

$$

Similarly for annihilation, the symmetric deletion $\vert\alpha\rangle \lhd_+ \vert\alpha\rangle^{\otimes n}$ removes $\vert\alpha\rangle$ from each of the $n$ positions (all giving $+$ overlap), producing $n\;\vert\alpha\rangle^{\otimes(n-1)}$:

$$

\hat{b}_\alpha \vert n_\alpha\rangle = \frac{n}{\sqrt{n \cdot n!}}\;\vert\alpha\rangle^{\otimes(n-1)} = \sqrt{n}\;\vert n-1\rangle

$$
:::

All occupation number states can be constructed from the vacuum:

$$

\vert n_\alpha\rangle = \frac{(\hat{b}^\dagger_\alpha)^{n_\alpha}}{\sqrt{n_\alpha!}}\;\vert 0_\alpha\rangle

$$ (eq-boson-fock-construction)

The **number operator** $\hat{n}_\alpha = \hat{b}^\dagger_\alpha \hat{b}_\alpha$ counts bosons: $\hat{n}_\alpha \vert n_\alpha\rangle = n_\alpha \vert n_\alpha\rangle$, with $n_\alpha = 0, 1, 2, \ldots$ unbounded.



#### Fermion Creation and Annihilation Operators

The **fermion creation operator** $\hat{c}^\dagger_\alpha$ and **annihilation operator** $\hat{c}_\alpha$ are defined analogously, using antisymmetric insertion and deletion:

:::{admonition} Definition: Fermion Creation and Annihilation
:class: important

Acting on an $N$-particle state $\vert\Psi\rangle$:

$$

\hat{c}^\dagger_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N+1}}\;\vert\alpha\rangle \rhd_- \vert\Psi\rangle, \qquad \hat{c}_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N}}\;\vert\alpha\rangle \lhd_- \vert\Psi\rangle

$$ (eq-fermion-op-def)

where $\rhd_-$ inserts antisymmetrically and $\lhd_-$ deletes antisymmetrically.
:::

:::{admonition} Example: $\hat{c}^\dagger_1 \vert 1_1, 1_2\rangle = 0$ (Pauli Exclusion)
:class: dropdown example

**Problem.** Attempt to create one more fermion in the already-occupied state $\vert 1\rangle$, starting from $\vert 1_1, 1_2\rangle = (\vert 1\rangle\vert 2\rangle - \vert 2\rangle\vert 1\rangle)/\sqrt{2}$.

**Solution.** Apply $\hat{c}^\dagger_1 = \frac{1}{\sqrt{3}}\vert 1\rangle \rhd_-$:

$$

\hat{c}^\dagger_1 \vert 1_1, 1_2\rangle = \frac{1}{\sqrt{2}} \left(\frac{1}{\sqrt{3}}\vert 1\rangle \rhd_- \vert 1\rangle\vert 2\rangle - \frac{1}{\sqrt{3}}\vert 1\rangle \rhd_- \vert 2\rangle\vert 1\rangle\right)

$$

Compute each antisymmetric insertion:

$$

\vert 1\rangle \rhd_- \vert 1\rangle\vert 2\rangle = \vert 1\rangle\vert 1\rangle\vert 2\rangle - \vert 1\rangle(\vert 1\rangle \rhd_- \vert 2\rangle) = \vert 1\rangle\vert 1\rangle\vert 2\rangle - \vert 1\rangle\vert 1\rangle\vert 2\rangle + \vert 1\rangle\vert 2\rangle\vert 1\rangle

$$

$$

\vert 1\rangle \rhd_- \vert 2\rangle\vert 1\rangle = \vert 1\rangle\vert 2\rangle\vert 1\rangle - \vert 2\rangle(\vert 1\rangle \rhd_- \vert 1\rangle) = \vert 1\rangle\vert 2\rangle\vert 1\rangle - \vert 2\rangle\vert 1\rangle\vert 1\rangle + \vert 2\rangle\vert 1\rangle\vert 1\rangle = \vert 1\rangle\vert 2\rangle\vert 1\rangle

$$

Sum: $\frac{1}{\sqrt{6}}(\vert 1\rangle\vert 2\rangle\vert 1\rangle - \vert 1\rangle\vert 2\rangle\vert 1\rangle) = 0$.

The antisymmetric insertion produces exact cancellation—no two fermions can occupy the same state.
:::

:::{admonition} Example: $\hat{c}_2 \vert 1_1, 1_2\rangle = -\vert 1_1, 0_2\rangle$ (Fermion Sign)
:class: dropdown example

**Problem.** Annihilate a fermion from state $\vert 2\rangle$, starting from $\vert 1_1, 1_2\rangle = (\vert 1\rangle\vert 2\rangle - \vert 2\rangle\vert 1\rangle)/\sqrt{2}$.

**Solution.** Apply $\hat{c}_2 = \frac{1}{\sqrt{2}}\vert 2\rangle \lhd_-$:

$$

\hat{c}_2 \vert 1_1, 1_2\rangle = \frac{1}{\sqrt{2}}\left(\frac{1}{\sqrt{2}}\vert 2\rangle \lhd_- \vert 1\rangle\vert 2\rangle - \frac{1}{\sqrt{2}}\vert 2\rangle \lhd_- \vert 2\rangle\vert 1\rangle\right)

$$

Compute each antisymmetric deletion:

$$

\vert 2\rangle \lhd_- \vert 1\rangle\vert 2\rangle = \delta_{21}\vert 2\rangle - \vert 1\rangle(\vert 2\rangle \lhd_- \vert 2\rangle) = 0 - \vert 1\rangle \cdot 1 = -\vert 1\rangle

$$

$$

\vert 2\rangle \lhd_- \vert 2\rangle\vert 1\rangle = \delta_{22}\vert 1\rangle - \vert 2\rangle(\vert 2\rangle \lhd_- \vert 1\rangle) = \vert 1\rangle - 0 = \vert 1\rangle

$$

Sum: $\frac{1}{2}(-\vert 1\rangle - \vert 1\rangle) = -\vert 1\rangle = -\vert 1_1, 0_2\rangle$.

The **minus sign** (the fermion sign) appears because removing $\vert 2\rangle$ requires "moving past" the occupied state $\vert 1\rangle$, picking up a factor of $(-1)$.
:::

**Action on single-mode Fock states.** Starting from the single-mode vacuum $\vert 0_\alpha\rangle = \mathbb{1}$ and applying the fermion creation operator, one can derive the action on any Fock state:

:::{admonition} Fermion Operators on Single-Mode Fock States
:class: important

$$

\hat{c}^\dagger_\alpha \vert 0_\alpha\rangle = \vert 1_\alpha\rangle, \qquad \hat{c}^\dagger_\alpha \vert 1_\alpha\rangle = 0

$$ (eq-fermion-creation-single)

$$

\hat{c}_\alpha \vert 1_\alpha\rangle = \vert 0_\alpha\rangle, \qquad \hat{c}_\alpha \vert 0_\alpha\rangle = 0

$$ (eq-fermion-annihilation-single)

If the state is empty, creation fills it. If already occupied, creation gives zero (Pauli exclusion). The vacuum is quenched by annihilation.
:::

:::{admonition} Derivation: Fermion Action on Single-Mode Fock States
:class: dropdown information

Apply the definitions
$\hat{c}^\dagger_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N+1}} \vert\alpha\rangle \rhd_- \vert\Psi\rangle$
and
$\hat{c}_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N}} \vert\alpha\rangle \lhd_- \vert\Psi\rangle$.

**(A) Creation action**

Starting from the vacuum:

$$

\hat{c}^\dagger_\alpha \vert 0_\alpha\rangle = \frac{1}{\sqrt{1}} \vert\alpha\rangle \rhd_- \mathbb{1} = \vert\alpha\rangle = \vert 1_\alpha\rangle

$$

On an already occupied state:

$$

\hat{c}^\dagger_\alpha \vert 1_\alpha\rangle = \frac{1}{\sqrt{2}} \vert\alpha\rangle \rhd_- \vert\alpha\rangle

$$

Using the recursive insertion rule:

$$

\vert\alpha\rangle \rhd_- \vert\alpha\rangle = \vert\alpha\rangle \otimes \vert\alpha\rangle - \vert\alpha\rangle \otimes (\vert\alpha\rangle \rhd_- \mathbb{1}) = \vert\alpha\rangle\vert\alpha\rangle - \vert\alpha\rangle\vert\alpha\rangle = 0

$$

So $\hat{c}^\dagger_\alpha \vert 1_\alpha\rangle = 0$ (Pauli exclusion).

**(B) Annihilation action**

On an occupied state:

$$

\hat{c}_\alpha \vert 1_\alpha\rangle = \frac{1}{\sqrt{1}} \vert\alpha\rangle \lhd_- \vert\alpha\rangle

$$

Using the recursive deletion rule with $\vert\alpha\rangle = \vert\alpha\rangle \otimes \mathbb{1}$:

$$

\vert\alpha\rangle \lhd_- \vert\alpha\rangle = \delta_{\alpha\alpha}\,\mathbb{1} - \vert\alpha\rangle \otimes (\vert\alpha\rangle \lhd_- \mathbb{1}) = 1 \cdot \mathbb{1} - 0 = \mathbb{1} = \vert 0_\alpha\rangle

$$

On the vacuum:

$$

\hat{c}_\alpha \vert 0_\alpha\rangle = \vert\alpha\rangle \lhd_- \mathbb{1} = 0

$$

by the base case of the deletion rule (no fermion to remove).
:::

The Fock state is constructed as $\vert n_\alpha\rangle = (\hat{c}^\dagger_\alpha)^{n_\alpha}\vert 0_\alpha\rangle$ with $n_\alpha \in \{0, 1\}$.

The **number operator** $\hat{n}_\alpha = \hat{c}^\dagger_\alpha \hat{c}_\alpha$ counts fermions: $\hat{n}_\alpha \vert n_\alpha\rangle = n_\alpha\vert n_\alpha\rangle$.



### Summary

- **Identical particles** have no labels; the many-body state $\vert\Psi\rangle$ must be symmetric (bosons) or antisymmetric (fermions) under particle exchange.
- **Symmetric states** are constructed by summing over all permutations of single-particle kets (the permanent); bosons allow multiple occupation.
- **Antisymmetric states** are constructed with alternating signs (the Slater determinant); Pauli exclusion ($n_\alpha \in \{0,1\}$) follows automatically.
- **Insertion and deletion rules** $\rhd_\pm$ and $\lhd_\pm$ define recursive procedures for adding/removing particles while preserving symmetry.
- **Boson operators**: $\hat{b}^\dagger_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N+1}}\vert\alpha\rangle \rhd_+ \vert\Psi\rangle$; on Fock states, $\hat{b}^\dagger\vert n\rangle = \sqrt{n+1}\,\vert n+1\rangle$, $\hat{b}\,\vert n\rangle = \sqrt{n}\,\vert n-1\rangle$.
- **Fermion operators**: $\hat{c}^\dagger_\alpha \vert\Psi\rangle = \frac{1}{\sqrt{N+1}}\vert\alpha\rangle \rhd_- \vert\Psi\rangle$; on Fock states, $\hat{c}^\dagger\vert 0\rangle = \vert 1\rangle$, $\hat{c}^\dagger\vert 1\rangle = 0$ (Pauli exclusion).
