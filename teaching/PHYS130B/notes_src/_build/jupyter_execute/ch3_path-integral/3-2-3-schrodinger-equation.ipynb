## Lecture Notes

:::{admonition} Definition
:class: important

The **Schrödinger equation** is the fundamental equation governing the time evolution of quantum systems. It emerges naturally from the path integral through a small-time-step approximation, revealing that the wavefunction's phase evolves according to the classical Hamilton-Jacobi equation while its amplitude spreads due to quantum diffusion—a beautiful synthesis of classical and quantum mechanics.
:::

:::{admonition} Discussion: Path Integral to Schrödinger
:class: dropdown tip

We derived Schrödinger equation from the path integral using the small-time-step limit. Yet traditionally, Schrödinger equation is *postulated* as a fundamental axiom. This raises subtle questions:

- The derivation assumes the path integral can be evaluated order-by-order in time-step $\epsilon$. Can this procedure always be justified rigorously, or are there cases where it breaks down?
- Starting from Schrödinger equation, can we uniquely reconstruct the path integral, or are there multiple path integral formulations that give the same differential equation?
- The classical action $S = \int L \, \mathrm{d}t$ appears in the path integral as $\mathrm{e}^{\mathrm{i}S/\hbar}$. Is this the *defining* property of $\hbar$, or a consequence of the commutation relations?
- For a time-dependent Hamiltonian $H(t)$ (e.g., external field switched on/off), how does the path integral formulation change? Can you write down the modified propagator formula?
:::
### The Kinetic Energy Term: Revisiting Time-Slicing

We showed that for a free particle, the time-sliced propagator over one small interval $\varepsilon$ is

$$K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) = \left(\frac{m}{2\pi \mathrm{i}\hbar\varepsilon}\right)^{d/2} \exp\left(\frac{\mathrm{i}m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon}\right)$$


Now we ask: **What Taylor expansion in $\varepsilon$ does this satisfy?**

#:::{admonition} Derivation: Time Derivative of Propagator
:class: dropdown note

Write

$$K_\varepsilon = N(\varepsilon) \exp(\mathrm{i}S_\varepsilon/\hbar)$$


where $N(\varepsilon) = (m/2\pi \mathrm{i}\hbar\varepsilon)^{d/2}$ is the normalization and $S_\varepsilon = \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon}$ is the action. Taking the time derivative:

$$\frac{\partial K_\varepsilon}{\partial\varepsilon} = K_\varepsilon \left(\frac{\partial (\ln N)}{\partial\varepsilon} + \frac{\mathrm{i}}{\hbar}\frac{\partial S_\varepsilon}{\partial\varepsilon}\right)$$


Computing the derivatives:

$$\frac{\partial (\ln N)}{\partial\varepsilon} = -\frac{\mathrm{d}}{2}\frac{1}{\varepsilon}$$


$$\frac{\partial S_\varepsilon}{\partial\varepsilon} = -\frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon^2}$$


So

$$\frac{\partial K_\varepsilon}{\partial\varepsilon} = K_\varepsilon \left(-\frac{\mathrm{d}}{2\varepsilon} - \frac{\mathrm{i}m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon^2}\right)$$


Rearranging:

$$\mathrm{i}\hbar \frac{\partial K_\varepsilon}{\partial\varepsilon} = K_\varepsilon \left(-\mathrm{i}\hbar\frac{\mathrm{d}}{2\varepsilon} + \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon^2}\right)$$


:::

This doesn't immediately look like a Schrödinger equation. We need to relate the spatial derivatives to the kinetic energy term.

#:::{admonition} Derivation: Laplacian of Free Propagator
:class: dropdown note

Compute the second spatial derivative with respect to $\boldsymbol{x}'$:

$$\nabla_{\boldsymbol{x}'}^2 K_\varepsilon = \nabla_{\boldsymbol{x}'}^2 \left[N(\varepsilon) \exp\left(\frac{\mathrm{i}m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon}\right)\right]$$


Using the product rule on $\mathrm{e}^{\mathrm{i}f(\boldsymbol{x}')}$ where $f = \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon}$:

$$\nabla_{\boldsymbol{x}'} \mathrm{e}^{\mathrm{i}f} = i(\nabla f) \mathrm{e}^{\mathrm{i}f} = i \frac{m(\boldsymbol{x}' - \boldsymbol{x})}{\hbar\varepsilon} \mathrm{e}^{\mathrm{i}f}$$


$$\nabla_{\boldsymbol{x}'}^2 \mathrm{e}^{\mathrm{i}f} = i\frac{m}{\hbar\varepsilon} \nabla^2(\boldsymbol{x}' - \boldsymbol{x}) \mathrm{e}^{\mathrm{i}f} + i^2 \frac{m^2(\boldsymbol{x}' - \boldsymbol{x})^2}{\hbar^2\varepsilon^2} \mathrm{e}^{\mathrm{i}f}$$


The Laplacian computation gives:

$$\nabla_{\boldsymbol{x}'}^2 \mathrm{e}^{\mathrm{i}f} = \frac{\mathrm{i}md}{\hbar\varepsilon} \mathrm{e}^{\mathrm{i}f} - \frac{m^2(\boldsymbol{x}' - \boldsymbol{x})^2}{\hbar^2\varepsilon^2} \mathrm{e}^{\mathrm{i}f}$$


Therefore:

$$\nabla_{\boldsymbol{x}'}^2 K_\varepsilon = K_\varepsilon \left(\frac{\mathrm{i}md}{\hbar\varepsilon} - \frac{m^2(\boldsymbol{x}' - \boldsymbol{x})^2}{\hbar^2\varepsilon^2}\right)$$


Rearranging:

$$-\frac{\hbar^2}{2m}\nabla_{\boldsymbol{x}'}^2 K_\varepsilon = K_\varepsilon \left(-\frac{\hbar d}{2\varepsilon} + \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon^2}\right)$$


:::

Comparing with the time derivative result above, we see that **in the limit $\varepsilon \to 0$**, the kinetic energy term dominates:

$$\mathrm{i}\hbar \frac{\partial K_\varepsilon}{\partial\varepsilon} \approx -\frac{\hbar^2}{2m}\nabla_{\boldsymbol{x}'}^2 K_\varepsilon + O(\varepsilon)$$


This is the **free-particle Schrödinger equation**!

### Adding Potential Energy: The Full Derivation

For a particle in a potential $V(\boldsymbol{x})$, the Lagrangian is $L = \frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})$. In a small time step $\varepsilon$:

$$S_\varepsilon = \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon} - V(\boldsymbol{x}) \varepsilon$$


For small $\varepsilon$, treating $V\varepsilon$ as a perturbation:

$$\mathrm{e}^{\mathrm{i}S_\varepsilon/\hbar} = \mathrm{e}^{\mathrm{i}m(\boldsymbol{x}'-\boldsymbol{x})^2/(2\hbar\varepsilon)} \mathrm{e}^{-\mathrm{i}V(\boldsymbol{x})\varepsilon/\hbar}$$


To leading order in $\varepsilon$:

$$\mathrm{e}^{-\mathrm{i}V(\boldsymbol{x})\varepsilon/\hbar} \approx 1 - \frac{\mathrm{i}V(\boldsymbol{x})\varepsilon}{\hbar}$$


So the single-slice propagator becomes

$$K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \approx K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \left(1 - \frac{\mathrm{i}V(\boldsymbol{x})\varepsilon}{\hbar}\right)$$


where $K_0$ is the free-particle propagator.

#:::{admonition} Derivation: Adding Potential to Time Evolution
:class: dropdown note

Now consider the wavefunction evolution from $t$ to $t + \varepsilon$:

$$\psi(\boldsymbol{x}', t + \varepsilon) = \int K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \, \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x}$$


$$\approx \int K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \left(1 - \frac{\mathrm{i}V(\boldsymbol{x})\varepsilon}{\hbar}\right) \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x}$$


$$= \int K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x} - \frac{\mathrm{i}\varepsilon}{\hbar} \int K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) V(\boldsymbol{x}) \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x}$$


The first term is the free-particle evolution. The second term, for small $\varepsilon$ and using $V(\boldsymbol{x}) \approx V(\boldsymbol{x}')$, gives:

$$\frac{\psi(\boldsymbol{x}', t + \varepsilon) - \psi(\boldsymbol{x}', t)}{\varepsilon} \approx \left(-\frac{\hbar^2}{2m}\nabla^2 - V(\boldsymbol{x}')\right) \psi(\boldsymbol{x}', t)$$


Taking the limit $\varepsilon \to 0$:

$$\frac{\partial\psi}{\partial t} = \frac{1}{\mathrm{i}\hbar}\left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right)\psi$$


:::

:::{admonition} The Schrödinger Equation
:class: important

$$\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right)\psi = \hat{H}\psi$$


This is the **time-dependent Schrödinger equation**. The path integral naturally generates it without additional assumptions.
:::

:::{admonition} Key Insight
:class: important

The path integral formulation is mathematically equivalent to the Schrödinger equation. Time-slicing, combined with Gaussian integrals and the continuum limit, derives the Schrödinger equation from first principles. This was Feynman's great insight: **quantum mechanics can be formulated as a sum over all possible histories, each weighted by the exponential of the classical action.**
:::

### The Time-Independent Schrödinger Equation

For stationary states with energy $E$, the wavefunction is $\psi(\boldsymbol{x}, t) = \phi(\boldsymbol{x}) \mathrm{e}^{-\mathrm{i}Et/\hbar}$. The propagator between two times at the same spatial point involves a phase factor:

$$\psi(\boldsymbol{x}, t_f) = \mathrm{e}^{-\mathrm{i}E(t_f - t_i)/\hbar} \psi(\boldsymbol{x}, t_i)$$


Energy eigenstates satisfy

$$E \phi(\boldsymbol{x}) = \hat{H} \phi(\boldsymbol{x}) = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right) \phi(\boldsymbol{x})$$


:::{admonition} Derivation: Schrödinger from Stationary Phase
:class: dropdown information

For stationary states, the wavefunction has the form $\psi(\boldsymbol{x}, t) = \phi(\boldsymbol{x}) \mathrm{e}^{-\mathrm{i}Et/\hbar}$. In the path integral formalism, the propagator from time $t_i$ to time $t_f$ is

$$K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int_{\boldsymbol{x}(t_i)=\boldsymbol{x}_i}^{\boldsymbol{x}(t_f)=\boldsymbol{x}_f} \mathrm{e}^{\mathrm{i}S[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]$$


For a stationary state with energy $E$ and evolution from $t_i = 0$ to $t_f = T$:

$$\psi(\boldsymbol{x}_f, T) = \int K(\boldsymbol{x}_f, T; \boldsymbol{x}_i, 0) \phi(\boldsymbol{x}_i) \, d^d\boldsymbol{x}_i = \mathrm{e}^{-\mathrm{i}ET/\hbar} \phi(\boldsymbol{x}_f)$$


The action along a classical path over time $T$ is $S = ET$. For stationary states, paths with action $S = ET$ contribute with **stationary phase**—small variations in the path produce negligible changes in the exponent.

For energy eigenstates, the path integral becomes dominated by paths that satisfy the stationary action principle at energy $E$. This is equivalent to requiring that $\phi(\boldsymbol{x})$ satisfies the **time-independent Schrödinger equation**:

$$E \phi(\boldsymbol{x}) = \hat{H}\phi(\boldsymbol{x})$$


:::

### Normalization and Probability Interpretation

The prefactor $(m/2\pi \mathrm{i}\hbar\varepsilon)^{d/2}$ in the propagator ensures proper normalization of probability. Specifically:

:::{admonition} Probability Conservation
:class: important

$$\int K(\boldsymbol{x}', t'; \boldsymbol{x}, t) \, d^d\boldsymbol{x} = 1 \quad \text{(for fixed } \boldsymbol{x}, t, t'\text{)}$$


This means that if $\int |\psi(\boldsymbol{x}, t)|^2 d^d\boldsymbol{x} = 1$ (normalized wavefunction), then

$$\int |\psi(\boldsymbol{x}, t')|^2 d^d\boldsymbol{x} = 1 \quad \text{(at later time)}$$


**Probability is conserved in time**—a fundamental requirement of quantum mechanics.
:::

:::{admonition} Note: Classical-Quantum Correspondence
:class: note

The transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the **correspondence principle**.

In quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::

### Summary

- **Time-slicing derivation**: Start with the path integral as a product of single-slice propagators; take the limit $\varepsilon \to 0$
- **Free particle**: The kinetic term in the action yields the Laplacian in the Schrödinger equation
- **Potential energy**: Added as an exponential factor; expands to $(1 - iV\varepsilon/\hbar)$ for small time steps
- **Full Schrödinger equation**: Emerges from the path integral without additional assumptions
- **Path integral and Schrödinger equivalence**: These two formulations are mathematically equivalent
- **Stationary states**: Energy eigenstates correspond to paths with stationary action for a given energy
- **Probability conservation**: The normalization factor ensures $\int |\psi|^2 = 1$ is preserved in time
