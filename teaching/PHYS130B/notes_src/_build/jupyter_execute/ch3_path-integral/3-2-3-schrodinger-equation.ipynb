## Lecture Notes

The Schrödinger equation emerges naturally from the path integral through a small-time-step approximation, revealing that the wavefunction's phase evolves according to the classical Hamilton-Jacobi equation while its amplitude spreads due to quantum diffusion—a beautiful synthesis of classical and quantum mechanics.

### The Kinetic Energy Term: Revisiting Time-SlicingWe showed that for a free particle, the time-sliced propagator over one small interval $\varepsilon$ is

$$K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) = \left(\frac{m}{2\pi i\hbar\varepsilon}\right)^{d/2} \exp\left(\frac{im(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon}\right)

$$ (eq-kinetic-propagator)

Now we ask: **What Taylor expansion in $\varepsilon$ does this satisfy?**

#### Expansion in Powers of $\varepsilon$:::{admonition} Derivation: Time Derivative of Propagator
:class: dropdown note

Write

$$K_\varepsilon = N(\varepsilon) \exp(iS_\varepsilon/\hbar)

$$where $N(\varepsilon) = (m/2\pi i\hbar\varepsilon)^{d/2}$ is the normalization and $S_\varepsilon = \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon}$ is the action.Taking the time derivative:

$$\frac{\partial K_\varepsilon}{\partial\varepsilon} = K_\varepsilon \left(\frac{\partial (\ln N)}{\partial\varepsilon} + \frac{i}{\hbar}\frac{\partial S_\varepsilon}{\partial\varepsilon}\right)

$$Computing the derivatives:

$$\frac{\partial (\ln N)}{\partial\varepsilon} = -\frac{d}{2}\frac{1}{\varepsilon}

$$$$\frac{\partial S_\varepsilon}{\partial\varepsilon} = -\frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon^2}

$$So

$$\frac{\partial K_\varepsilon}{\partial\varepsilon} = K_\varepsilon \left(-\frac{d}{2\varepsilon} - \frac{im(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon^2}\right)

$$ (eq-propagator-time-deriv)

Rearranging:

$$i\hbar \frac{\partial K_\varepsilon}{\partial\varepsilon} = K_\varepsilon \left(-i\hbar\frac{d}{2\varepsilon} + \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon^2}\right)

$$ (eq-ih-derivative)



:::

Hmm, this doesn't immediately look like a Schrödinger equation. We need to relate the spatial derivatives to the kinetic energy term.

#### Spatial Derivatives of the Kinetic Propagator:::{admonition} Derivation: Laplacian of Free Propagator
:class: dropdown note

Compute the second spatial derivative with respect to $\boldsymbol{x}'$:

$$\nabla_{\boldsymbol{x}'}^2 K_\varepsilon = \nabla_{\boldsymbol{x}'}^2 \left[N(\varepsilon) \exp\left(\frac{im(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon}\right)\right]

$$Using the product rule on $e^{if(\boldsymbol{x}')}$ where $f = \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\hbar\varepsilon}$:

$$\nabla_{\boldsymbol{x}'} e^{if} = i(\nabla f) e^{if} = i \frac{m(\boldsymbol{x}' - \boldsymbol{x})}{\hbar\varepsilon} e^{if}

$$$$\nabla_{\boldsymbol{x}'}^2 e^{if} = i\frac{m}{\hbar\varepsilon} \nabla^2(\boldsymbol{x}' - \boldsymbol{x}) e^{if} + i^2 \frac{m^2(\boldsymbol{x}' - \boldsymbol{x})^2}{\hbar^2\varepsilon^2} e^{if}

$$The Laplacian of $\boldsymbol{x}' - \boldsymbol{x}$ (with respect to $\boldsymbol{x}'$) is $d$ times the delta function in the distributional sense, but for smooth functions:

$$\nabla_{\boldsymbol{x}'}^2(\boldsymbol{x}' - \boldsymbol{x}) = 2d \text{ (coordinate sum)}

$$Actually, more carefully, $\sum_{i} \frac{\partial^2}{\partial x'_i} (x'_i - x_i) = d$, so

$$\nabla_{\boldsymbol{x}'}^2 e^{if} = \frac{imd}{\hbar\varepsilon} e^{if} - \frac{m^2(\boldsymbol{x}' - \boldsymbol{x})^2}{\hbar^2\varepsilon^2} e^{if}

$$Therefore:

$$\nabla_{\boldsymbol{x}'}^2 K_\varepsilon = K_\varepsilon \left(\frac{imd}{\hbar\varepsilon} - \frac{m^2(\boldsymbol{x}' - \boldsymbol{x})^2}{\hbar^2\varepsilon^2}\right)

$$ (eq-laplacian-kinetic)

Rearranging:

$$-\frac{\hbar^2}{2m}\nabla_{\boldsymbol{x}'}^2 K_\varepsilon = K_\varepsilon \left(-\frac{\hbar d}{2\varepsilon} + \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon^2}\right)

$$ (eq-kinetic-laplacian)

Comparing with Eq. (eq-ih-derivative)

:::

, we see that **in the limit $\varepsilon \to 0$**, the kinetic energy term dominates:

$$i\hbar \frac{\partial K_\varepsilon}{\partial\varepsilon} \approx -\frac{\hbar^2}{2m}\nabla_{\boldsymbol{x}'}^2 K_\varepsilon + O(\varepsilon)

$$ (eq-schrodinger-free)

This is the **free-particle Schrödinger equation**!

### Adding Potential Energy: The Full DerivationFor a particle in a potential $V(\boldsymbol{x})$, the Lagrangian is $L = \frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})$. In a small time step $\varepsilon$:

$$S_\varepsilon = \frac{m(\boldsymbol{x}' - \boldsymbol{x})^2}{2\varepsilon} - V(\boldsymbol{x}) \varepsilon

$$ (eq-action-with-potential)

For small $\varepsilon$, we can approximate $V(\boldsymbol{x}) \approx V(\boldsymbol{x}')$ or use the midpoint. More carefully, treating $V\varepsilon$ as a perturbation:

$$e^{iS_\varepsilon/\hbar} = e^{im(\boldsymbol{x}'-\boldsymbol{x})^2/(2\hbar\varepsilon)} e^{-iV(\boldsymbol{x})\varepsilon/\hbar}

$$ (eq-exponential-factored)

To leading order in $\varepsilon$:

$$e^{-iV(\boldsymbol{x})\varepsilon/\hbar} \approx 1 - \frac{iV(\boldsymbol{x})\varepsilon}{\hbar}

$$ (eq-potential-expansion)

So the single-slice propagator becomes

$$K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \approx K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \left(1 - \frac{iV(\boldsymbol{x})\varepsilon}{\hbar}\right)

$$ (eq-single-slice-with-potential)

where $K_0$ is the free-particle propagator.

#### Full Time Evolution:::{admonition} Derivation: Adding Potential to Time Evolution
:class: dropdown note

Now consider the wavefunction evolution from $t$ to $t + \varepsilon$:

$$\psi(\boldsymbol{x}', t + \varepsilon) = \int K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \, \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x}

$$$$\approx \int K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \left(1 - \frac{iV(\boldsymbol{x})\varepsilon}{\hbar}\right) \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x}

$$$$= \int K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x} - \frac{i\varepsilon}{\hbar} \int K_0(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) V(\boldsymbol{x}) \psi(\boldsymbol{x}, t) \, d^d\boldsymbol{x}

$$The first term is the free-particle evolution. The second term, for small $\varepsilon$ and using $V(\boldsymbol{x}) \approx V(\boldsymbol{x}')$, gives:

$$\frac{\psi(\boldsymbol{x}', t + \varepsilon) - \psi(\boldsymbol{x}', t)}{\varepsilon} \approx \left(-\frac{\hbar^2}{2m}\nabla^2 - V(\boldsymbol{x}')\right) \psi(\boldsymbol{x}', t)

$$ (eq-finite-diff-schrodinger)

Taking the limit $\varepsilon \to 0$:

:::



$$\frac{\partial\psi}{\partial t} = \frac{1}{i\hbar}\left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right)\psi

$$Or equivalently:

$$i\hbar\frac{\partial\psi}{\partial t} = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right)\psi

$$ (eq-schrodinger-full)

This is the **time-dependent Schrödinger equation**. The path integral naturally generates it.

:::{admonition} Key insight: Path integral reproduces Schrödinger equation:class: importantThe path integral formulation is not just an alternative formulation of quantum mechanics—it is **mathematically equivalent** to the Schrödinger equation. The time-slicing approximation, combined with Gaussian integrals and a careful limit, derives the Schrödinger equation from scratch.This equivalence was Feynman's great insight: **quantum mechanics can be formulated as a sum over all possible histories, each weighted by the classical action.**
:::
### The Time-Independent Schrödinger EquationFor stationary states with energy $E$, the wavefunction is $\psi(\boldsymbol{x}, t) = \phi(\boldsymbol{x}) e^{-iEt/\hbar}$. The propagator between two times at the same spatial point involves a phase factor:

$$\psi(\boldsymbol{x}, t_f) = e^{-iE(t_f - t_i)/\hbar} \psi(\boldsymbol{x}, t_i)

$$ (eq-stationary-phase)

More generally, energy eigenstates satisfy

$$E \phi(\boldsymbol{x}) = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right) \phi(\boldsymbol{x})

$$ (eq-time-independent-schrodinger)

This can also be derived from the path integral by looking for solutions where paths with action $S = Et$ give constructive interference—i.e., by applying the stationary phase condition to the path integral.

### Normalization and Probability InterpretationThe prefactor $(m/2\pi i\hbar\varepsilon)^{d/2}$ in the propagator ensures proper normalization of probability. Specifically:

$$\int K(\boldsymbol{x}', t'; \boldsymbol{x}, t) \, d^d\boldsymbol{x} = 1 \text{ (for fixed } \boldsymbol{x}, t, t'\text{)}

$$ (eq-propagator-normalization)

This means that if $\int |\psi(\boldsymbol{x}, t)|^2 d^d\boldsymbol{x} = 1$ (normalized wavefunction), then

$$\int |\psi(\boldsymbol{x}, t')|^2 d^d\boldsymbol{x} = 1 \text{ (at later time)}

$$ (eq-norm-conservation)

Probability is **conserved** in time—a fundamental requirement of quantum mechanics.

:::{admonition} Classical-quantum correspondence:class: seealsoThe transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the correspondence principle.
:::
:::{admonition} Action and phase:class: noteIn quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.


:::{admonition} Derivation: Time-Independent Schrödinger Equation from Stationary Phase
:class: dropdown note

For stationary states, the wavefunction has the form $\psi(\boldsymbol{x}, t) = \phi(\boldsymbol{x}) e^{-iEt/\hbar}$. In the path integral formalism, the propagator from time $t_i$ to time $t_f$ is

$$K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int_{\boldsymbol{x}(t_i)=\boldsymbol{x}_i}^{\boldsymbol{x}(t_f)=\boldsymbol{x}_f} e^{iS[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$

For a stationary state with energy $E$ and considering evolution from initial state $\phi(\boldsymbol{x}_i)$ at $t_i = 0$ to time $t_f = T$:

$$\psi(\boldsymbol{x}_f, T) = \int K(\boldsymbol{x}_f, T; \boldsymbol{x}_i, 0) \phi(\boldsymbol{x}_i) \, d^d\boldsymbol{x}_i = e^{-iET/\hbar} \phi(\boldsymbol{x}_f)

$$

The action along a classical path over time $T$ is $S = ET$. For stationary states, paths with action $S = ET$ contribute with **stationary phase**—small variations in the path produce negligible changes in the exponent.

For energy eigenstates, the path integral becomes dominated by paths that satisfy the **stationary action principle** at energy $E$:

$$\delta(S - ET) = 0 \quad \Rightarrow \quad \text{Euler-Lagrange equations with constraint } E = \text{const}

$$

This is equivalent to requiring that $\phi(\boldsymbol{x})$ satisfies the **time-independent Schrödinger equation**:

$$E \phi(\boldsymbol{x}) = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right) \phi(\boldsymbol{x})

$$

**Derivation from composition property**: Alternatively, for short time intervals $\varepsilon$, the stationary-state condition is

$$\phi(\boldsymbol{x}, t+\varepsilon) = e^{-iE\varepsilon/\hbar} \phi(\boldsymbol{x}, t)

$$

Expanding $e^{-iE\varepsilon/\hbar} \approx 1 - iE\varepsilon/\hbar$ and using the composition of propagators from the time-dependent Schrödinger equation, we obtain

$$E \phi(\boldsymbol{x}) = H \phi(\boldsymbol{x}) = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(\boldsymbol{x})\right) \phi(\boldsymbol{x})

$$

Thus, energy eigenstates are **eigenvalues of the Hamiltonian**, derived directly from the stationary-phase condition in the path integral.

:::

### Summary
- **Time-slicing derivation**: Start with the path integral written as a product of single-slice propagators; take the limit $\varepsilon \to 0$.
- **Free particle**: The kinetic term in the action, when differentiated, yields the Laplacian in the Schrödinger equation.
- **Potential energy**: Added as an exponential factor in the action; expands to $(1 - iV\varepsilon/\hbar)$ for small time steps.
- **Full Schrödinger equation**: Emerges from the path integral without additional assumptions. The path integral and Schrödinger equation are equivalent.
- **Stationary states**: Energy eigenstates correspond to paths with stationary action for a given energy.
- **Probability conservation**: The normalization factor ensures that $\int |\psi|^2 = 1$ is preserved in time.
