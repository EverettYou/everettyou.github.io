## Lecture Notes

:::{admonition} Definition
:class: important

**Instantons** are tunneling paths in imaginary time—real, finite-action solutions connecting distinct classical minima separated by barriers. They provide the dominant contribution to tunneling rates, explaining how quantum systems can explore classically forbidden regions. Instantons are central to both quantum tunneling and non-perturbative phenomena in field theory.
:::

### Tunneling in Imaginary Time

One of the most powerful applications of the imaginary-time path integral is understanding **tunneling** through classical barriers.

#### The Instanton Idea

In real time, a particle cannot classically traverse a potential barrier (no real path exists with $E < V$). But in **imaginary time**, the Euclidean action $S_E = \int (\frac{m}{2}\dot{x}^2 + V) d\tau$ is always positive. A path connecting two minima separated by a barrier can have finite, stationary action.

**Instanton**: An **instanton** is a **real solution** of the **Euclidean equations of motion** that connects two different classical equilibria:

$$m\ddot{x} = \frac{\mathrm{d}V}{dx}$$


These are **real trajectories** in imaginary time, not oscillations. An instanton solution approaches equilibrium at $\tau \to -\infty$ and $\tau \to +\infty$.

:::{admonition} Note: Not a Quantum State
:class: note

Instantons are classical solutions in imaginary time—they are not quantum states. Rather, they represent a particular way a particle can "tunnel" through a barrier in the imaginary-time formulation. When analytically continued back to real time, they describe the penetration probability.
:::

:::{admonition} Discussion: Instantons and Real-Time Physics
:class: dropdown tip

The instanton is a real, finite-action solution to the Euclidean (imaginary-time) equations of motion. Yet the quantum tunneling effect is a real-time phenomenon. This seems paradoxical:

- How does a solution in imaginary time (where $t = \mathrm{i}\tau$) encode information about real-time tunneling probability? Is the Wick rotation a mere mathematical trick, or does it reveal a deep symmetry?
- The semi-classical approximation weights saddle points of the action. The instanton is a saddle point of the Euclidean action. Does this explain why instantons dominate the imaginary-time path integral?
- In the real-time path integral, the WKB approximation uses barrier penetration (real paths that classically forbidden). In the imaginary-time picture, the instanton smoothly crosses the barrier (like rolling over a hill in imaginary time). Do these describe the same physics from different angles?
- For a double-well potential, the instanton action determines tunneling rate via $\Gamma \propto \mathrm{e}^{-S_{\text{inst}}/\hbar}$. How does this compare quantitatively to the WKB tunneling exponent $\Gamma \propto \mathrm{e}^{-2\int p \, \mathrm{d}x/\hbar}$?
:::

### The Double-Well Potential

Consider the **symmetric double-well potential**:

$$V(x) = \lambda(x^2 - a^2)^2$$


This has two minima at $x = \pm a$ with $V(\pm a) = 0$ and a barrier of height $V(0) = \lambda a^4$ at the origin.

#### Ground State Splitting

Quantum mechanically, the ground state is **split** into two levels:
- **Symmetric state** $|+\rangle$: $\psi$ even, symmetric around the origin
- **Antisymmetric state** $|-\rangle$: $\psi$ odd

:::{admonition} Energy Splitting from Tunneling
:class: important

The **energy splitting** is

$$\Delta E = E_+ - E_- = 2|\langle + | \text{tunneling} | - \rangle|$$


where "tunneling" represents the matrix element connecting the two wells.
:::

#### Instanton Solution

In imaginary time, there exists a kink solution connecting $x(-\infty) = -a$ to $x(+\infty) = +a$. This is the **instanton**:

$$x_{\text{inst}}(\tau) = a \tanh\left(\omega \frac{\tau - \tau_0}{2}\right)$$


where $\omega$ depends on the potential parameters and $\tau_0$ is the center position.

:::{admonition} Instanton Action and Tunneling
:class: important

The **action** of the instanton is

$$S_{\text{inst}} = \int_{-\infty}^{+\infty} \left(\frac{m}{2}\dot{x}^2 + V(x)\right) d\tau$$


For the quartic double-well, this evaluates to

$$S_{\text{inst}} = \frac{C}{\lambda^{1/4}} a^3 m^{1/2}$$


where $C$ is a numerical factor.
:::

:::{admonition} Derivation: Euclidean Equations of Motion
:class: dropdown information

The Euclidean action is

$$S_E[x] = \int \left(\frac{m}{2}\dot{x}^2 + V(x)\right) d\tau$$


Varying with respect to $x(\tau)$:

$$\frac{\delta S_E}{\delta x(\tau)} = 0 \quad \Rightarrow \quad m\ddot{x} = \frac{\mathrm{d}V}{dx}$$


This is the **equation of motion in imaginary time**. Unlike the real-time equation $m\ddot{x} = -dV/dx$ (with a minus sign), the Euclidean equation has the kinetic energy and potential energy terms both contributing with the same sign. This allows a real solution (the instanton) that connects two minima while climbing the potential.

For the double-well, the kink solution $x_{\text{inst}}(\tau) = a \tanh(\omega\tau/2)$ satisfies this equation exactly.

:::

### Tunneling Rate from Instanton

In the imaginary-time path integral, the tunneling amplitude is related to the instanton contribution:

$$\text{Tunneling amplitude} \propto \mathrm{e}^{-S_{\text{inst}}/\hbar}$$


The **decay rate** (for an excited state above the barrier) is

$$\Gamma = A \mathrm{e}^{-2S_{\text{inst}}/\hbar}$$


The factor of 2 appears because tunneling can occur in either direction (over and back).

#### Instanton Gas Approximation

More rigorously, the partition function includes:
- **Zero-instanton** sector: paths staying in one well
- **One-instanton** sector: paths making one transition
- **Two-instanton** sector: paths making a round trip
- **Many-instanton** sectors: multiple transitions

The **instanton gas approximation** sums the first few terms, giving the energy splitting:

$$\Delta E = 2\hbar \mathrm{e}^{-S_{\text{inst}}/\hbar} \times \text{(pre-exponential factor)}$$


### Comparison: WKB vs. Instanton

Both WKB and the instanton method give tunneling rates, but via different routes:

| Method | Formulation | Key Quantity | Result |
|--------|-------------|--------------|--------|
| **WKB** | Real time | Decay of $\psi$ in forbidden region | $T \propto \exp(-2\int\kappa dx/\hbar)$ |
| **Instanton** | Imaginary time | Stationary action on kink path | $\Delta E \propto \exp(-S_{\text{inst}}/\hbar)$ |

For the double-well, both give similar exponential suppression, but the instanton method is more systematic for multi-dimensional problems and field theories.

:::{admonition} Discussion: WKB vs. Instantons
:class: dropdown tip

For tunneling through a potential barrier, both WKB and the instanton method give semi-classical predictions, but via different routes:

- For a symmetric double-well potential, compute the tunneling rate using both WKB (real-time barrier penetration integral) and instantons (Euclidean action). Do they give the same exponent? If not, where does the discrepancy arise?
- WKB is an expansion in $\hbar$; the instanton contribution is non-perturbative (all orders in $\hbar$). Yet both predict exponential suppression at small $\hbar$. How can a non-perturbative result match the $\hbar \to 0$ behavior of a perturbative expansion?
- The instanton formalism naturally includes ALL instanton-anti-instanton pairs. In real time, does WKB account for bouncing effects (successive barriers), or must it be extended?
- In quantum field theory, instantons predict proton decay and baryon number violation. Does the real-time WKB picture have a natural quantum field theory counterpart?
:::

### The Instanton-Anti-Instanton Pair

A **two-instanton** configuration (instanton followed by anti-instanton) represents a **round trip**: left well $\to$ barrier $\to$ right well $\to$ barrier $\to$ left well.

The action of an instanton-anti-instanton pair separated by imaginary time $T$ is approximately

$$S_{2\text{-inst}} = 2S_{\text{inst}} + (\text{interaction energy}) \times T$$


For large separation, the pair contributions oscillate at the energy splitting frequency: $\Delta E/\hbar \propto \mathrm{e}^{-S_{\text{inst}}/\hbar}$.

### Physical Interpretation

#### Why Instantons Work

The imaginary-time path integral naturally includes all tunneling paths. The **dominant contribution** comes from paths that are closest to classical (stationary action). An instanton is a classical solution in imaginary time—it's a saddle point of the Euclidean action. By the stationary phase approximation, it dominates the path integral around the tunneling rate.

#### Bounce Solution

In field theory, a **bounce** is an instanton followed by an anti-instanton, representing a field configuration that "bounces" over the barrier and back. The tunneling probability $\propto \exp(-S_{\text{bounce}}/\hbar)$ decays with the action of the bounce—a semiclassical result for decay through a potential barrier in quantum field theory.

### Connection to Real Time

To recover real-time tunneling probabilities from imaginary-time instantons, one performs **analytic continuation** $\tau \to it$:

$$x_{\text{inst}}(\tau) = a\tanh(\omega\tau/2) \quad \Rightarrow \quad x(t) = a\tan(\omega t/2) \quad (\text{complex trajectory})$$


The resulting **complex trajectory** in real time describes the wave penetration into the classically forbidden region and encodes the tunneling amplitude.

:::{admonition} Note: Classical-Quantum Correspondence
:class: note

The transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the **correspondence principle**.

In quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::

### Summary

- **Instanton**: Real solution of Euclidean equations of motion, connecting two wells in imaginary time
- **Kink solution**: For double-well, the instanton is a smooth interpolation $x(\tau) = a\tanh(...)$ between minima
- **Euclidean action**: $S_{\text{inst}} = \int(\frac{m}{2}\dot{x}^2 + V) d\tau$ determines tunneling rate exponentially
- **Tunneling amplitude**: $\propto \exp(-S_{\text{inst}}/\hbar)$; more precise than WKB in many cases
- **Energy splitting**: For double-well, ground state doublet split by $\Delta E \propto \exp(-S_{\text{inst}}/\hbar)$
- **Instanton gas**: Higher-order terms (multiple instantons) refine the calculation
- **Analytic continuation**: Instantons connect imaginary-time saddle points to real-time penetration amplitudes
- **Field theory**: Instantons are crucial for understanding tunneling in QCD, electroweak theory, and non-perturbative effects
