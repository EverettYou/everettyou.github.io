## Lecture Notes

Instantons are tunneling paths in imaginary time—real, finite-action solutions connecting distinct classical minima separated by barriers. They provide the dominant contribution to tunneling rates, explaining how quantum systems can explore classically forbidden regions. Instantons are central to both quantum tunneling and non-perturbative phenomena in field theory.

### Tunneling in Imaginary TimeOne of the most powerful applications of the imaginary-time path integral is understanding **tunneling** through classical barriers.

#### The Instanton IdeaIn real time, a particle cannot classically traverse a potential barrier (no real path exists with $E < V$). But in **imaginary time**, the Euclidean action $S_E = \int (\frac{m}{2}\dot{x}^2 + V) d\tau$ is always positive. A path connecting two minima separated by a barrier can have finite, stationary action.**Instanton**: An **instanton** is a **real solution** of the **Euclidean equations of motion** that connects two different classical equilibria:

$$\frac{\delta S_E}{\delta x(\tau)} = 0 \quad \Rightarrow \quad m\ddot{x} = \frac{dV}{dx}

$$ (eq-euclidean-eom)

These are **real trajectories** in imaginary time, not oscillations. An instanton solution approaches equilibrium at $\tau \to -\infty$ and $\tau \to +\infty$.

:::{admonition} Not a quantum state:class: noteInstantons are classical solutions in imaginary time—they are not quantum states. Rather, they represent a particular way a particle can "tunnel" through a barrier in the imaginary-time formulation. When analytically continued back to real time, they describe the penetration probability.
:::
### The Double-Well PotentialConsider the **symmetric double-well potential**:

$$V(x) = \lambda(x^2 - a^2)^2

$$ (eq-double-well)

This has two minima at $x = \pm a$ with $V(\pm a) = 0$ and a barrier of height $V(0) = \lambda a^4$ at the origin.

#### Ground State SplittingQuantum mechanically, the ground state is **split** into two levels:
- **Symmetric state** $|+\rangle$: $\psi$ even, symmetric around the origin.
- **Antisymmetric state** $|-\rangle$: $\psi$ odd.The **energy splitting** is

$$\Delta E = E_+ - E_- = 2|\langle + | \text{tunneling} | - \rangle|

$$ (eq-energy-splitting)

where "tunneling" represents the matrix element connecting the two wells.

#### Instanton SolutionIn imaginary time, there exists a kink solution connecting $x(-\infty) = -a$ to $x(+\infty) = +a$. This is the **instanton**:

$$x_{\text{inst}}(\tau) = a \tanh\left(\omega \frac{\tau - \tau_0}{2}\right)

$$ (eq-kink-instanton)

where $\omega$ depends on the potential parameters and $\tau_0$ is the center position.The **action** of the instanton is

$$S_{\text{inst}} = \int_{-\infty}^{+\infty} \left(\frac{m}{2}\dot{x}^2 + V(x)\right) d\tau

$$ (eq-instanton-action)

For the quartic double-well, this evaluates to

$$S_{\text{inst}} = \frac{C}{\lambda^{1/4}} a^3 m^{1/2}

$$ (eq-instanton-action-value)

where $C$ is a numerical factor.

### Tunneling Rate from InstantonIn the imaginary-time path integral, the tunneling amplitude is related to the instanton contribution:

$$\text{Tunneling amplitude} \propto e^{-S_{\text{inst}}/\hbar}

$$ (eq-tunneling-amplitude)

The **decay rate** (for an excited state above the barrier) is

$$\Gamma = A e^{-2S_{\text{inst}}/\hbar}

$$ (eq-decay-rate)

The factor of 2 appears because tunneling can occur in either direction (over and back).

#### Instanton Gas ApproximationMore rigorously, the partition function includes:
- **Zero-instanton** sector: paths staying in one well.
- **One-instanton** sector: paths making one transition.
- **Two-instanton** sector: paths making a round trip.
- **Many-instanton** sectors: multiple transitions.The **instanton gas approximation** sums the first few terms, giving the energy splitting:

$$\Delta E = 2\hbar e^{-S_{\text{inst}}/\hbar} \times \text{(pre-exponential factor)}

$$ (eq-splitting-instanton)

### Comparison: WKB vs. InstantonBoth WKB and the instanton method give tunneling rates, but via different routes:| Method | Formulation | Key Quantity | Result ||--------|-------------|--------------|--------|| **WKB** | Real time | Decay of $\psi$ in forbidden region | $T \propto \exp(-2\int\kappa dx/\hbar)$ || **Instanton** | Imaginary time | Stationary action on kink path | $\Delta E \propto \exp(-S_{\text{inst}}/\hbar)$ |For the double-well, both give similar exponential suppression, but the instanton method is more systematic for multi-dimensional problems and field theories.

### The Instanton-Anti-Instanton PairA **two-instanton** configuration (instanton followed by anti-instanton) represents a **round trip**: left well $\to$ barrier $\to$ right well $\to$ barrier $\to$ left well.The action of an instanton-anti-instanton pair separated by imaginary time $T$ is approximately

$$S_{2-\text{inst}} = 2S_{\text{inst}} + \text{(interaction energy)} \times T

$$ (eq-pair-action)

For large separation, the pair contributions oscillate at the energy splitting frequency: $\Delta E/\hbar \propto e^{-S_{\text{inst}}/\hbar}$.

### Physical Interpretation

#### Why Instantons WorkThe imaginary-time path integral naturally includes all tunneling paths. The **dominant contribution** comes from paths that are closest to classical (stationary action). An instanton is a classical solution in imaginary time—it's a saddle point of the Euclidean action. By the stationary phase approximation, it dominates the path integral around the tunneling rate.

#### Bounce SolutionIn field theory, a **bounce** is an instanton followed by an anti-instanton, representing a field configuration that "bounces" over the barrier and back. The tunneling probability $\propto \exp(-S_{\text{bounce}}/\hbar)$ decays with the action of the bounce—a semiclassical result for decay through a potential barrier in quantum field theory.

### Connection to Real TimeTo recover real-time tunneling probabilities from imaginary-time instantons, one performs **analytic continuation** $\tau \to it$:

$$x_{\text{inst}}(\tau) = a\tanh(\omega\tau/2) \quad \Rightarrow \quad x(t) = a\tan(\omega t/2) \quad (\text{complex trajectory})

$$The resulting **complex trajectory** in real time describes the wave penetration into the classically forbidden region and encodes the tunneling amplitude.

:::{admonition} Classical-quantum correspondence:class: seealsoThe transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the correspondence principle.
:::
:::{admonition} Action and phase:class: noteIn quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::
### Summary
- **Instanton**: Real solution of Euclidean equations of motion, connecting two wells in imaginary time.
- **Kink solution**: For double-well, the instanton is a smooth interpolation $x(\tau) = a\tanh(...)$ between minima.
- **Euclidean action**: $S_{\text{inst}} = \int(\frac{m}{2}\dot{x}^2 + V) d\tau$ determines tunneling rate exponentially.
- **Tunneling amplitude**: $\propto \exp(-S_{\text{inst}}/\hbar)$; more precise than WKB in many cases.
- **Energy splitting**: For double-well, ground state doublet split by $\Delta E \propto \exp(-S_{\text{inst}}/\hbar)$.
- **Instanton gas**: Higher-order terms (multiple instantons) refine the calculation.
- **Analytic continuation**: Instantons connect imaginary-time saddle points to real-time penetration amplitudes.
- **Field theory**: Instantons are crucial for understanding tunneling in QCD, electroweak theory, and non-perturbative effects.
