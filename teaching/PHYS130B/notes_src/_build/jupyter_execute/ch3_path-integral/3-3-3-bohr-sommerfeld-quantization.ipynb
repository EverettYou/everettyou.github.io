# 3.3.3 Bohr-Sommerfeld Quantization

:::{admonition} Prompts

- Apply Bohr-Sommerfeld to the harmonic oscillator — why is the result exact?

## Lecture Notes

### Overview

**Bohr-Sommerfeld quantization** applies the WKB approximation to bound states. Single-valuedness of the wavefunction after a complete classical orbit requires $\oint p\,\mathrm{d}x = 2\pi\hbar(n + \tfrac{1}{2})$. The $\tfrac{1}{2}$ arises from $\pi/4$ phase shifts at each turning point — the **Maslov index** — and explains zero-point energy. For the harmonic oscillator, this gives the exact spectrum $E_n = (n+\tfrac{1}{2})\hbar\omega$. Geometrically, quantization is **constructive interference of periodic orbits**.

### Derivation from WKB

For a bound state oscillating between turning points $x_a$ and $x_b$ (where $E = V(x)$), the WKB wavefunction must return to itself after one complete orbit.

The phase accumulated going from $x_a$ to $x_b$ is $\frac{1}{\hbar}\int_{x_a}^{x_b} p(x)\,\mathrm{d}x$. Each turning point contributes an additional $\pi/4$ phase shift from the connection formulas (§3.3.2). The total round-trip phase is

$$
\Phi_\mathrm{total} = \frac{2}{\hbar}\int_{x_a}^{x_b} p(x)\,\mathrm{d}x + 2 \times \frac{\pi}{4} + 2 \times \frac{\pi}{4} = \frac{1}{\hbar}\oint p\,\mathrm{d}x + \pi
$$

Single-valuedness requires $\Phi_\mathrm{total} = 2\pi n$, giving:

:::{admonition} Bohr-Sommerfeld Quantization Rule
:class: important

$$
\oint p(x)\,\mathrm{d}x = 2\pi\hbar\left(n + \frac{1}{2}\right), \quad n = 0, 1, 2, \ldots
$$

where $p(x) = \sqrt{2m(E - V(x))}$ and the integral is over the full classical orbit. The $\tfrac{1}{2}$ is the **Maslov correction** from turning-point phase shifts.
:::

:::{admonition} Derivation: Round-Trip Phase and the Maslov Index
:class: dropdown information

We derive the Bohr-Sommerfeld condition by carefully tracking the WKB wavefunction through one complete classical orbit between turning points $x_a$ and $x_b$.

**Step 1: WKB wavefunction in the classically allowed region.**
Between $x_a$ and $x_b$, the WKB solution is a superposition of right- and left-moving waves:

$$
\psi(x) = \frac{A}{\sqrt{p(x)}} \mathrm{e}^{\mathrm{i}\Phi(x)/\hbar} + \frac{B}{\sqrt{p(x)}} \mathrm{e}^{-\mathrm{i}\Phi(x)/\hbar}, \quad \Phi(x) = \int_{x_a}^{x} p(x')\,\mathrm{d}x'
$$

**Step 2: Connection formula at the right turning point $x_b$.**
Near $x_b$, the potential is approximately linear: $E - V(x) \approx -V'(x_b)(x - x_b)$. The connection formula (from Airy function matching, §3.3.2) requires the WKB solution to match onto an exponentially decaying solution for $x > x_b$. This imposes:

$$
\psi(x) \propto \frac{1}{\sqrt{p(x)}} \cos\left(\frac{1}{\hbar}\int_{x}^{x_b} p(x')\,\mathrm{d}x' - \frac{\pi}{4}\right), \quad x \text{ near } x_b
$$

Rewriting in terms of $\Phi(x) = \int_{x_a}^x p\,\mathrm{d}x'$, we define $\Phi_{\mathrm{total}} = \int_{x_a}^{x_b} p\,\mathrm{d}x'$, so

$$
\psi(x) \propto \frac{1}{\sqrt{p(x)}} \cos\left(\frac{\Phi_{\mathrm{total}} - \Phi(x)}{\hbar} - \frac{\pi}{4}\right)
$$

**Step 3: Connection formula at the left turning point $x_a$.**
Similarly, matching onto an exponentially decaying solution for $x < x_a$ requires:

$$
\psi(x) \propto \frac{1}{\sqrt{p(x)}} \cos\left(\frac{\Phi(x)}{\hbar} - \frac{\pi}{4}\right), \quad x \text{ near } x_a
$$

**Step 4: Consistency condition.**
Both expressions must represent the same wavefunction. Expanding the cosines:

$$
\cos\left(\frac{\Phi}{\hbar} - \frac{\pi}{4}\right) \propto \cos\left(\frac{\Phi_{\mathrm{total}} - \Phi}{\hbar} - \frac{\pi}{4}\right)
$$

This holds if and only if

$$
\frac{\Phi}{\hbar} - \frac{\pi}{4} = \pm\left(\frac{\Phi_{\mathrm{total}} - \Phi}{\hbar} - \frac{\pi}{4}\right) + 2\pi n
$$

Taking the $+$ sign: $2\Phi/\hbar = \Phi_{\mathrm{total}}/\hbar + 2\pi n$, which must hold for all $x$ — impossible. Taking the $-$ sign:

$$
\frac{\Phi_{\mathrm{total}}}{\hbar} - \frac{\pi}{2} = 2\pi n \implies \int_{x_a}^{x_b} p\,\mathrm{d}x = \pi\hbar\left(2n + 1\right) = \pi\hbar(2n+1)
$$

**Step 5: Convert to orbit integral.**
The full orbit integral counts the forward and return paths: $\oint p\,\mathrm{d}x = 2\int_{x_a}^{x_b} p\,\mathrm{d}x$. Thus:

$$
\oint p\,\mathrm{d}x = 2\pi\hbar\left(n + \frac{1}{2}\right)
$$

The $\tfrac{1}{2}$ (the **Maslov index** $\mu/2$ with $\mu = 2$ for two turning points) arises from the two $\pi/4$ phase shifts — one at each turning point. This is a topological contribution: each turning point where the classical momentum vanishes contributes $+1$ to the Maslov index.
:::

:::{admonition} Discussion: Semiclassical Quantization
:class: dropdown tip

The Bohr-Sommerfeld condition looks almost classical — an integral over a closed orbit. Yet it predicts quantum energy levels. This raises foundational questions:

- The condition works for hydrogen but fails for helium. What changes when there are two interacting electrons?
- The $\tfrac{1}{2}$ gives zero-point energy. Is this purely quantum, or does it have a wave-optics analogue?
- For chaotic (non-integrable) systems, there are no closed action variables. How does quantum mechanics handle the spectrum when semiclassical intuition breaks down?
:::

### Application: Harmonic Oscillator

For $V(x) = \frac{1}{2}m\omega^2 x^2$, the turning points are $x = \pm a$ with $a = \sqrt{2E/(m\omega^2)}$.

:::{admonition} Example: Harmonic Oscillator Spectrum
:class: example

The momentum is $p(x) = m\omega\sqrt{a^2 - x^2}$. The orbit integral:

$$
\oint p\,\mathrm{d}x = 2\int_{-a}^{a} m\omega\sqrt{a^2 - x^2}\,\mathrm{d}x = \pi m\omega a^2 = \frac{2\pi E}{\omega}
$$

Applying Bohr-Sommerfeld:

$$
\frac{2\pi E_n}{\omega} = 2\pi\hbar\left(n + \frac{1}{2}\right) \implies E_n = \hbar\omega\left(n + \frac{1}{2}\right)
$$

This is **exactly the quantum result**. The harmonic oscillator is special because its quadratic potential makes WKB exact — all higher-order corrections vanish identically.
:::

:::{admonition} Derivation: Harmonic Oscillator Orbit Integral
:class: dropdown information

We evaluate $\oint p\,\mathrm{d}x$ for $V(x) = \frac{1}{2}m\omega^2 x^2$ at energy $E$.

**Turning points.** Setting $E = V(x)$ gives $x_{\pm} = \pm a$ where $a = \sqrt{2E/(m\omega^2)}$.

**Momentum.** In the allowed region $|x| < a$:

$$
p(x) = \sqrt{2m(E - \tfrac{1}{2}m\omega^2 x^2)} = m\omega\sqrt{a^2 - x^2}
$$

**Orbit integral.** By symmetry, $\oint = 2\int_{-a}^{a}$:

$$
\oint p\,\mathrm{d}x = 2m\omega \int_{-a}^{a} \sqrt{a^2 - x^2}\,\mathrm{d}x
$$

Substitute $x = a\sin\theta$, $\mathrm{d}x = a\cos\theta\,\mathrm{d}\theta$, with $\theta \in [-\pi/2, \pi/2]$:

$$
= 2m\omega a^2 \int_{-\pi/2}^{\pi/2} \cos^2\theta\,\mathrm{d}\theta = 2m\omega a^2 \cdot \frac{\pi}{2} = \pi m\omega a^2
$$

**Substituting $a^2 = 2E/(m\omega^2)$:**

$$
\oint p\,\mathrm{d}x = \pi m\omega \cdot \frac{2E}{m\omega^2} = \frac{2\pi E}{\omega}
$$

**Applying Bohr-Sommerfeld:** $\frac{2\pi E_n}{\omega} = 2\pi\hbar(n + \tfrac{1}{2})$, giving

$$
E_n = \hbar\omega\left(n + \frac{1}{2}\right)
$$

**Why is this exact?** The harmonic oscillator potential is purely quadratic, so the WKB approximation $p(x) = \sqrt{2m(E-V)}$ involves no approximation beyond the leading semiclassical order. Formally, the path integral for a quadratic action is Gaussian and can be evaluated exactly; all corrections beyond leading order in $\hbar$ vanish. This is equivalent to the statement that the semiclassical propagator is exact for quadratic Hamiltonians.
:::

### Application: Hydrogen Atom

For the Coulomb potential $V(r) = -e^2/(4\pi\epsilon_0 r)$, combining angular momentum quantization ($L = n_\phi\hbar$) with radial Bohr-Sommerfeld quantization reproduces the Bohr energy levels:

$$
E_n = -\frac{13.6\;\mathrm{eV}}{n^2}, \quad n = 1, 2, 3, \ldots
$$

The result is exact due to a special cancellation of WKB errors in the Coulomb potential. For general potentials, Bohr-Sommerfeld is an excellent approximation that improves with increasing quantum number $n$.

:::{admonition} EBK Quantization
:class: dropdown note

For systems with $N$ degrees of freedom that are **integrable** (possessing $N$ independent constants of motion), the Einstein-Brillouin-Keller (EBK) generalization quantizes each action variable separately:

$$
J_i = \oint p_i\,\mathrm{d}q_i = 2\pi\hbar(n_i + \alpha_i)
$$

where $\alpha_i$ is the Maslov index for each degree of freedom. For the 2D isotropic oscillator, this gives $E = \hbar\omega(n_x + n_y + 1)$, exactly matching the quantum result. EBK fails for non-integrable (chaotic) systems where action variables do not exist.
:::

:::{admonition} Derivation: Bohr-Sommerfeld Energy Levels for Hydrogen
:class: dropdown information

We derive $E_n = -13.6\,\text{eV}/n^2$ by quantizing the radial and angular actions of the Kepler orbit.

**Setup.** Write the classical energy for hydrogen with $V(r) = -e^2/(4\pi\epsilon_0 r)$ (setting $e^2 \equiv e^2/(4\pi\epsilon_0)$ for brevity):

$$
E = \frac{p_r^2}{2m} + \frac{L^2}{2mr^2} - \frac{e^2}{r}
$$

**Step 1: Angular action quantization.**

The angular coordinate $\phi$ runs $[0, 2\pi)$, so:

$$
J_\phi = \oint p_\phi\,\mathrm{d}\phi = 2\pi L = 2\pi n_\phi \hbar \implies L = n_\phi\hbar, \quad n_\phi = 0, 1, 2, \ldots
$$

with no Maslov correction (periodic orbit, no turning point).

**Step 2: Radial action integral.**

The radial action for $E < 0$ is evaluated by contour integration. Factoring the integrand:

$$
J_r = \oint p_r\,\mathrm{d}r = \oint \sqrt{-p_\infty^2 + \frac{2me^2}{r} - \frac{L^2}{r^2}}\,\mathrm{d}r
$$

where $p_\infty^2 = -2mE > 0$. This integral has two branch-point singularities at $r_1, r_2$ (turning points) and a pole at $r = 0$. Using the residue theorem with a keyhole contour around the branch cut $[r_1, r_2]$ plus contributions from the poles at $r = 0$ and $r \to \infty$:

$$
J_r = 2\pi\left(\frac{me^2}{p_\infty} - L\right) = 2\pi\left(e^2\sqrt{\frac{m}{-2E}} - L\right)
$$

**Step 3: Apply quantization with Maslov index.**

The radial motion has two classical turning points $r_1$ (inner) and $r_2$ (outer), giving Maslov contribution $\tfrac{1}{2} + \tfrac{1}{2} = 1$. The quantization condition is:

$$
J_r = 2\pi\hbar(n_r + 1), \quad n_r = 0, 1, 2, \ldots
$$

Substituting:

$$
e^2\sqrt{\frac{m}{-2E}} - n_\phi\hbar = (n_r + 1)\hbar
$$

$$
e^2\sqrt{\frac{m}{-2E}} = (n_r + n_\phi + 1)\hbar \equiv n\hbar
$$

where $n = n_r + n_\phi + 1 \geq 1$ is the **principal quantum number**.

**Step 4: Solve for $E_n$.**

Square both sides:

$$
e^4 \cdot \frac{m}{-2E_n} = n^2\hbar^2 \implies E_n = -\frac{me^4}{2n^2\hbar^2}
$$

In SI units with $e^2 \to e^2/(4\pi\epsilon_0)$:

$$
E_n = -\frac{me^4}{2(4\pi\epsilon_0)^2\hbar^2} \cdot \frac{1}{n^2} = -\frac{13.6\,\text{eV}}{n^2} \quad \checkmark
$$

**Why the result is exact.** For most potentials, the WKB approximation introduces errors of order $\hbar^2$ that shift the energy levels. For the Coulomb potential, the radial and angular Maslov corrections conspire to cancel these errors exactly, because the Coulomb problem has an extra conserved quantity (the Laplace-Runge-Lenz vector) making it "maximally superintegrable."
:::

### Connection to the Path Integral

Bohr-Sommerfeld quantization is a direct consequence of the path integral. The propagator for a bound state receives contributions from periodic orbits — paths that close on themselves after time $T$. The condition for **constructive interference** among these periodic orbits is precisely $\oint p\,\mathrm{d}x = 2\pi\hbar(n + \tfrac{1}{2})$.

:::{admonition} Quantization = Constructive Interference
:class: attention

Discrete energy levels arise because only certain energies produce periodic orbits whose accumulated phase is a multiple of $2\pi$. All other energies lead to destructive interference and are forbidden.
:::

:::{admonition} Derivation: Bohr-Sommerfeld from the Path Integral
:class: dropdown information

We show how the Bohr-Sommerfeld condition emerges as a constructive interference condition for periodic orbits in the path integral.

**Step 1: Trace formula.**
The density of states $\rho(E) = \sum_n \delta(E - E_n)$ can be extracted from the trace of the propagator:

$$
\rho(E) = -\frac{1}{\pi}\mathrm{Im}\,\mathrm{Tr}\left(\frac{1}{E - \hat{H} + \mathrm{i}\epsilon}\right) = -\frac{1}{\pi}\mathrm{Im} \int_0^\infty \frac{\mathrm{d}T}{\hbar}\, \mathrm{e}^{\mathrm{i}ET/\hbar} \,\mathrm{Tr}\left(\mathrm{e}^{-\mathrm{i}\hat{H}T/\hbar}\right)
$$

**Step 2: Semiclassical trace.**
In the path integral, $\mathrm{Tr}(\mathrm{e}^{-\mathrm{i}\hat{H}T/\hbar}) = \int \mathrm{d}x\, K(x,T;x,0)$. The stationary phase approximation selects **periodic orbits** — classical paths that return to their starting point after time $T$. Each periodic orbit $\gamma$ with period $T_\gamma$ and action $S_\gamma = \oint_\gamma p\,\mathrm{d}x - E T_\gamma$ contributes a phase $\mathrm{e}^{\mathrm{i}S_\gamma/\hbar}$.

**Step 3: Constructive interference.**
For a 1D bound system with a single primitive orbit of period $T(E)$ and action $S(E) = \oint p\,\mathrm{d}x$, the $r$-th repetition has action $rS(E)$ and picks up an additional Maslov phase $r\mu\pi/2$ (where $\mu = 2$ for two turning points). The trace becomes a geometric series:

$$
\rho_{\mathrm{sc}}(E) \sim \mathrm{Re} \sum_{r=1}^{\infty} \mathrm{e}^{\mathrm{i}r(S/\hbar - \mu\pi/2)} = \mathrm{Re} \sum_{r=1}^{\infty} \mathrm{e}^{\mathrm{i}r(S/\hbar - \pi)}
$$

This sum produces **delta-function peaks** when the exponent is a multiple of $2\pi$:

$$
\frac{S(E)}{\hbar} - \pi = 2\pi n \implies \oint p\,\mathrm{d}x = 2\pi\hbar\left(n + \frac{1}{2}\right)
$$

This is precisely the Bohr-Sommerfeld condition. Quantized energy levels correspond to energies where all repetitions of the periodic orbit interfere constructively.
:::

This completes the semiclassical arc of Chapter 3: the path integral (§3.2) sums over all paths; stationary phase (§3.3.1) selects the classical path; WKB (§3.3.2) builds semiclassical wavefunctions; Bohr-Sommerfeld extracts discrete spectra from the constructive interference condition.

:::{admonition} Derivation: Bohr-Sommerfeld Applied to the Infinite Square Well
:class: dropdown information

The infinite square well ($V = 0$ for $0 < x < L$, $V = \infty$ elsewhere) is exactly soluble and serves as a clean test of Bohr-Sommerfeld quantization.

**Classical orbit.** A particle of energy $E$ bounces between the walls. The momentum is $p = \sqrt{2mE}$ (constant in the interior). The orbit consists of a right-moving stroke ($0 \to L$) and a left-moving stroke ($L \to 0$), each contributing momentum $+p$ and $-p$.

**Action integral.** The phase-space orbit is a rectangle in the $(x, p)$ plane. The enclosed area is:

$$
\oint p\, \mathrm{d}x = 2pL = 2L\sqrt{2mE}.
$$

**Maslov index.** At each hard wall, the WKB wavefunction undergoes a **full $\pi$ phase shift** (not the $\pi/2$ of a soft turning point). A hard wall forces the wavefunction to zero: $\psi(0) = \psi(L) = 0$. The boundary condition is equivalent to two reflections with total phase $\pi$ each, contributing a Maslov index $\mu = 2 \times 2 = 4$ (not the usual $\mu = 2$ for soft turning points).

**Quantization condition.** With total phase shift $2\pi$ (two hard-wall reflections, each $\pi$):

$$
\oint p\,\mathrm{d}x - \mu \cdot \frac{\pi\hbar}{2} = 2\pi n\hbar, \quad n = 1, 2, 3, \ldots
$$

$$
2L\sqrt{2mE_n} - 4 \cdot \frac{\pi\hbar}{2} = 2\pi n\hbar.
$$

$$
\sqrt{2mE_n} = \frac{\pi\hbar n}{L} \quad \Longrightarrow \quad E_n = \frac{\pi^2 \hbar^2 n^2}{2mL^2}.
$$

This is the **exact quantum result**. The Bohr-Sommerfeld approximation is exact here because (1) the potential is piecewise constant (no higher-order WKB corrections), and (2) the boundary conditions are hard walls, whose Maslov phase is integer rather than fractional.

**Contrast with soft turning points.** For a harmonic oscillator or hydrogen atom, the turning points are soft (the potential passes through $E$ with finite slope), giving $\pi/2$ phase shifts and $\mu = 2$. The quantization becomes $\oint p\, \mathrm{d}x = 2\pi\hbar(n + 1/2)$, yielding the zero-point energy $\hbar\omega/2$ that is absent in the square well.
:::

### Summary

- **Bohr-Sommerfeld rule**: $\oint p\,\mathrm{d}x = 2\pi\hbar(n+\tfrac{1}{2})$ quantizes bound-state energies from single-valuedness of the wavefunction
- **Maslov index**: the $\tfrac{1}{2}$ comes from $\pi/4$ phase shifts at each turning point, giving zero-point energy
- **Exact for harmonic oscillator**: $E_n = \hbar\omega(n+\tfrac{1}{2})$; also exact for hydrogen by special cancellation
- **Path integral origin**: quantization = constructive interference of periodic orbits, closing the semiclassical arc of Chapter 3
