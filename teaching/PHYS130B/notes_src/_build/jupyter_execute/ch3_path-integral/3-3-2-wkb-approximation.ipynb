## Lecture Notes

### Overview

The **WKB (Wentzel-Kramers-Brillouin) approximation** solves the Schrödinger equation semiclassically when the potential varies slowly compared to the de Broglie wavelength. The ansatz $\psi = A\,\mathrm{e}^{\mathrm{i}S/\hbar}$ splits the wavefunction into a rapidly oscillating phase and a slowly varying amplitude. At leading order in $\hbar$, the phase satisfies the **Hamilton-Jacobi equation** — classical mechanics — while the amplitude is fixed by probability conservation.

### The WKB Ansatz

Write the wavefunction as

$$
\psi(x, t) = A(x, t) \, \mathrm{e}^{\mathrm{i}S(x, t)/\hbar}
$$

Substituting into the Schrödinger equation and expanding in powers of $\hbar$:

:::{admonition} WKB Hierarchy
:class: important

- **Leading order** ($\hbar^0$): the phase $S$ satisfies the **Hamilton-Jacobi equation**

$$
\frac{\partial S}{\partial t} + \frac{1}{2m}\left(\frac{\partial S}{\partial x}\right)^2 + V(x) = 0
$$

- **Next order** ($\hbar^1$): the amplitude $A$ satisfies a **continuity equation** — probability conservation.
:::

:::{admonition} Derivation: WKB Equations from Schrödinger
:class: dropdown information

Substitute $\psi = A\,\mathrm{e}^{\mathrm{i}S/\hbar}$ into the Schrödinger equation $\mathrm{i}\hbar\partial_t\psi = -\frac{\hbar^2}{2m}\partial_x^2\psi + V\psi$. We compute each piece separately.

**Time derivative:**

$$
\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = \mathrm{i}\hbar\left(\dot{A} + \frac{\mathrm{i}}{\hbar}A\dot{S}\right)\mathrm{e}^{\mathrm{i}S/\hbar} = \left(\mathrm{i}\hbar\dot{A} - A\dot{S}\right)\mathrm{e}^{\mathrm{i}S/\hbar}
$$

**Spatial derivatives:**

$$
\frac{\partial\psi}{\partial x} = \left(A' + \frac{\mathrm{i}}{\hbar}AS'\right)\mathrm{e}^{\mathrm{i}S/\hbar}
$$

$$
\frac{\partial^2\psi}{\partial x^2} = \left(A'' + \frac{2\mathrm{i}}{\hbar}A'S' + \frac{\mathrm{i}}{\hbar}AS'' - \frac{A}{\hbar^2}(S')^2\right)\mathrm{e}^{\mathrm{i}S/\hbar}
$$

**Substituting** and canceling $\mathrm{e}^{\mathrm{i}S/\hbar}$:

$$
\mathrm{i}\hbar\dot{A} - A\dot{S} = -\frac{\hbar^2}{2m}\left(A'' + \frac{2\mathrm{i}}{\hbar}A'S' + \frac{\mathrm{i}}{\hbar}AS'' - \frac{A}{\hbar^2}(S')^2\right) + VA
$$

Rearranging and sorting by powers of $\hbar$:

$$
A\left[\dot{S} + \frac{(S')^2}{2m} + V\right] + \mathrm{i}\hbar\left[-\dot{A} - \frac{A'S'}{m} - \frac{AS''}{2m}\right] + \hbar^2\left[\frac{A''}{2m}\right] = 0
$$

**Order $\hbar^0$:** Setting the leading (real) bracket to zero gives the **Hamilton-Jacobi equation**:

$$
\frac{\partial S}{\partial t} + \frac{1}{2m}\left(\frac{\partial S}{\partial x}\right)^2 + V(x) = 0
$$

**Order $\hbar^1$:** Setting the next bracket to zero:

$$
\dot{A} + \frac{1}{m}A'S' + \frac{A}{2m}S'' = 0
$$

Multiplying by $2A$ and recognizing $\partial_x(A^2 S'/m) = 2AA'S'/m + A^2 S''/m$:

$$
\frac{\partial(A^2)}{\partial t} + \frac{\partial}{\partial x}\left(\frac{A^2}{m}\frac{\partial S}{\partial x}\right) = 0
$$

This is a **continuity equation** for $\rho = A^2$ with velocity $v = S'/m = p/m$, confirming probability conservation. The $\hbar^2$ term $A''/2m$ is dropped — this is the WKB approximation.
:::

### Stationary States: Allowed and Forbidden Regions

For energy eigenstates with $S(x,t) = S_0(x) - Et$, the Hamilton-Jacobi equation gives $S_0'(x) = \pm p(x)$ where $p(x) = \sqrt{2m(E - V(x))}$ is the local classical momentum.

:::{admonition} WKB Wavefunctions
:class: important

**Classically allowed** ($E > V$, real momentum):

$$
\psi(x) \sim \frac{1}{\sqrt{p(x)}} \exp\left(\pm\frac{\mathrm{i}}{\hbar}\int^x p(x')\,\mathrm{d}x'\right)
$$

**Classically forbidden** ($E < V$, imaginary momentum $p \to \mathrm{i}\kappa$):

$$
\psi(x) \sim \frac{1}{\sqrt{\kappa(x)}} \exp\left(\mp\frac{1}{\hbar}\int^x \kappa(x')\,\mathrm{d}x'\right)
$$

where $\kappa(x) = \sqrt{2m(V(x) - E)}$.
:::

The amplitude $1/\sqrt{p}$ reflects **probability conservation**: where the particle moves slowly (small $p$), it spends more time, so the probability density is larger.

:::{admonition} Derivation: WKB Amplitude from Probability Conservation
:class: dropdown information

For stationary states $S(x,t) = S_0(x) - Et$, we have $A = A(x)$ only. The continuity equation $\partial_t(A^2) + \partial_x(A^2 p/m) = 0$ reduces to

$$
\frac{\mathrm{d}}{\mathrm{d}x}\left(A^2(x)\frac{p(x)}{m}\right) = 0
$$

since $A^2$ is time-independent. Therefore $A^2 p/m = \text{const}$, giving

$$
A(x) = \frac{C}{\sqrt{p(x)}}
$$

for some constant $C$. The **physical meaning** is transparent: the probability current $j = A^2 v = A^2 p/m$ must be spatially uniform in a stationary state (no probability accumulation anywhere). Where the particle moves slowly ($p$ small), $A^2$ must be large to maintain constant current — hence $A \propto 1/\sqrt{p}$.

In the forbidden region, $p(x) \to \mathrm{i}\kappa(x)$ with $\kappa = \sqrt{2m(V-E)}$, and the same argument gives $A \propto 1/\sqrt{\kappa}$.
:::

### Connection Formulas at Turning Points

At a **turning point** $x_0$ where $E = V(x_0)$, the momentum vanishes and the WKB solutions diverge. The exact solution near $x_0$ (an Airy function) bridges the two regions:

**Allowed side** ($x < x_0$):

$$
\psi \sim \frac{C}{\sqrt{p}} \sin\left(\frac{1}{\hbar}\int_x^{x_0} p\,\mathrm{d}x' + \frac{\pi}{4}\right)
$$

**Forbidden side** ($x > x_0$):

$$
\psi \sim \frac{C}{\sqrt{\kappa}} \exp\left(-\frac{1}{\hbar}\int_{x_0}^x \kappa\,\mathrm{d}x'\right)
$$

The $\pi/4$ phase shift is essential — it accounts for the smooth transition through the turning point and determines the Maslov index in quantization conditions (§3.3.3).


:::{admonition} Discussion: WKB Exactness
:class: dropdown tip

WKB is an approximation, yet it gives exact energy levels for the harmonic oscillator. This raises questions:

- For which potentials is WKB exact? What special property must they have?
- WKB predicts tunneling rates via $T \propto \mathrm{e}^{-2\gamma}$. Can you design a potential where WKB gives a very wrong tunneling rate?
- Why does WKB work well for atomic energy levels when $\hbar$ is not truly small? Is there a hidden small parameter?
:::

### Tunneling

For a particle incident on a barrier where $V > E$ between turning points $a$ and $b$:

:::{admonition} WKB Tunneling Probability
:class: important

$$
T \approx \mathrm{e}^{-2\gamma}, \quad \gamma = \frac{1}{\hbar}\int_a^b \kappa(x)\,\mathrm{d}x = \frac{1}{\hbar}\int_a^b \sqrt{2m(V(x) - E)}\,\mathrm{d}x
$$

:::

The tunneling probability is exponentially sensitive to the barrier width and height. The penetration depth into the forbidden region is $\lambda_\mathrm{decay} = \hbar/\kappa$.

:::{admonition} Derivation: WKB Tunneling Formula
:class: dropdown information

Consider a potential barrier with two turning points $a$ and $b$ where $V(a) = V(b) = E$, and $V > E$ for $a < x < b$. We apply the connection formulas at each turning point.

**Step 1: Set up the three regions.**

Region I ($x < a$, allowed): incoming + reflected wave

$$
\psi_\mathrm{I} = \frac{A}{\sqrt{p}}\mathrm{e}^{+\frac{\mathrm{i}}{\hbar}\int_a^x p\,\mathrm{d}x'} + \frac{B}{\sqrt{p}}\mathrm{e}^{-\frac{\mathrm{i}}{\hbar}\int_a^x p\,\mathrm{d}x'}
$$

Region II ($a < x < b$, forbidden): growing + decaying

$$
\psi_\mathrm{II} = \frac{C}{\sqrt{\kappa}}\mathrm{e}^{-\frac{1}{\hbar}\int_a^x \kappa\,\mathrm{d}x'} + \frac{D}{\sqrt{\kappa}}\mathrm{e}^{+\frac{1}{\hbar}\int_a^x \kappa\,\mathrm{d}x'}
$$

Region III ($x > b$, allowed): transmitted wave only

$$
\psi_\mathrm{III} = \frac{F}{\sqrt{p}}\mathrm{e}^{+\frac{\mathrm{i}}{\hbar}\int_b^x p\,\mathrm{d}x'}
$$

**Step 2: Apply connection formulas.** At turning point $b$, the outgoing wave in Region III connects through the Airy-function matching to the forbidden region, selecting the decaying exponential. The connection formula at $b$ relates $F$ to the Region II amplitudes evaluated at $x = b$. The growing exponential $D\,\mathrm{e}^{+\gamma}$ (where $\gamma = \frac{1}{\hbar}\int_a^b\kappa\,\mathrm{d}x'$) connects to a wave of amplitude $\sim D\,\mathrm{e}^{\gamma}$ on the allowed side, while the decaying term $C\,\mathrm{e}^{-\gamma}$ is exponentially suppressed.

**Step 3: Transmission coefficient.** In the limit of a thick barrier ($\gamma \gg 1$), the dominant contribution to the transmitted amplitude comes from the exponentially decaying piece. Tracing through both connection formulas:

$$
\left\lvert\frac{F}{A}\right\rvert^2 \approx \mathrm{e}^{-2\gamma}
$$

where

$$
\gamma = \frac{1}{\hbar}\int_a^b \kappa(x)\,\mathrm{d}x = \frac{1}{\hbar}\int_a^b\sqrt{2m(V(x) - E)}\,\mathrm{d}x
$$

The transmission probability $T = |F/A|^2 \approx \mathrm{e}^{-2\gamma}$ is exponentially suppressed by the "action" accumulated in the classically forbidden region. This result also follows directly from the path integral: tunneling paths traverse the barrier in imaginary time, contributing $\mathrm{e}^{-S_E/\hbar}$ where $S_E$ is the Euclidean action (see §3.4.3 on instantons).
:::

:::{admonition} Example: Electron Tunneling
:class: dropdown example

**Problem.** Estimate the tunneling probability for an electron through a 1 nm barrier with height 1 eV above its energy.

**Solution.** $\gamma \approx (10^{-9}\;\mathrm{m})\sqrt{2(9.1\times 10^{-31}\;\mathrm{kg})(1.6\times 10^{-19}\;\mathrm{J})}/(10^{-34}\;\mathrm{J{\cdot}s}) \approx 5$, giving $T \sim \mathrm{e}^{-10} \approx 10^{-4.3}$. Rare but non-negligible at atomic scales — this is why quantum tunneling drives processes like alpha decay and scanning tunneling microscopy.
:::

### Validity of WKB

The WKB approximation requires the potential to vary slowly on the scale of the local de Broglie wavelength:

$$
\left|\frac{\mathrm{d}\ln p}{\mathrm{d}x}\right| = \left|\frac{V'(x)}{2(E - V(x))}\right| \ll 1
$$

This breaks down at turning points (remedied by connection formulas), potential discontinuities, and rapidly oscillating potentials.

:::{admonition} Derivation: WKB Validity Criterion
:class: dropdown information

The WKB approximation drops the $\hbar^2$ term $A''/(2m)$ relative to the leading-order terms. For self-consistency, we need this dropped term to be much smaller than the retained terms. Since $A \propto 1/\sqrt{p}$:

$$
A = \frac{C}{\sqrt{p}}, \quad A' = -\frac{C\,p'}{2p^{3/2}}, \quad A'' = -\frac{C\,p''}{2p^{3/2}} + \frac{3C\,(p')^2}{4p^{5/2}}
$$

The ratio of the dropped $\hbar^2$ term to the leading kinetic term $p^2/(2m)$ is of order

$$
\frac{\hbar^2 A''/(2mA)}{p^2/(2m)} \sim \frac{\hbar^2}{p^2}\left(\frac{p''}{p} - \frac{3(p')^2}{2p^2}\right)
$$

For this to be small, we need (keeping the dominant piece):

$$
\hbar\left\lvert\frac{p'}{p^2}\right\rvert \ll 1 \implies \left\lvert\frac{\mathrm{d}p^{-1}}{\mathrm{d}x}\right\rvert \ll \frac{1}{\hbar} \implies \left\lvert\frac{\mathrm{d}\lambda}{\mathrm{d}x}\right\rvert \ll 2\pi
$$

where $\lambda = 2\pi\hbar/p$ is the local de Broglie wavelength. In words: **the wavelength must change by much less than itself over one wavelength** — the potential must be slowly varying on the scale of $\lambda$.

Using $p' = -mV'/(p)$ and $p = \sqrt{2m(E-V)}$, this becomes

$$
\left\lvert\frac{\mathrm{d}\ln p}{\mathrm{d}x}\right\rvert = \frac{m\lvert V'\rvert}{p^2} = \frac{\lvert V'\rvert}{2(E - V)} \ll \frac{p}{\hbar} = \frac{1}{\lambda/(2\pi)}
$$

This diverges at turning points where $p \to 0$, confirming that the connection formula treatment is essential there.
:::

### Summary

- **WKB ansatz** $\psi = A\,\mathrm{e}^{\mathrm{i}S/\hbar}$: phase obeys Hamilton-Jacobi (classical mechanics); amplitude fixed by probability conservation
- **Connection formulas**: match oscillating and decaying solutions across turning points with $\pi/4$ phase shifts
- **Tunneling**: $T \approx \mathrm{e}^{-2\gamma}$ with $\gamma = \hbar^{-1}\int \kappa\,\mathrm{d}x$ — exponentially suppressed by barrier width and height
- **Validity**: excellent for smooth, slowly varying potentials; breaks down at turning points and sharp features
