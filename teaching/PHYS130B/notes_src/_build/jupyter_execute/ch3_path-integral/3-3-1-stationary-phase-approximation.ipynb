## Lecture Notes

The stationary phase approximation reveals the deep connection between classical and quantum mechanics: when $\hbar$ is small, oscillatory integrals are dominated by points where the phase is stationary. This principle underlies the emergence of classical trajectories from quantum path integrals and explains why the classical path is so special.

### The Stationary Phase Approximation (SPA)

Many integrals in quantum mechanics have the form

$$

I = \int e^{i\Phi(x)/\hbar} f(x) \, dx

$$ (eq-phase-integral)

where $\Phi(x)$ is a **phase function**, $f(x)$ is a slowly-varying **amplitude**, and $\hbar$ is small. When $\hbar$ is tiny, the integrand oscillates wildly unless $\Phi$ is slowly varying.

:::{admonition} Stationary Phase Approximation
:class: important

The integral $I = \int e^{i\Phi(x)/\hbar} f(x) \, dx$ is dominated by points where the phase is **stationary**:

$$

\frac{d\Phi}{dx}\bigg|_{x_0} = 0

$$ (eq-stationary-condition)

Near a stationary point $x_0$, expand the phase:

$$

\Phi(x) \approx \Phi(x_0) + \frac{1}{2}\Phi''(x_0)(x - x_0)^2 + \cdots

$$ (eq-phase-expansion)

Then:

$$

I \approx e^{i\Phi(x_0)/\hbar} f(x_0) \int e^{i\Phi''(x_0)(x-x_0)^2/(2\hbar)} dx

$$

Using the Gaussian integral formula $\int e^{iax^2} dx = \sqrt{\pi/(ia)}$:

$$

I \approx e^{i\Phi(x_0)/\hbar} f(x_0) \sqrt{\frac{2\pi\hbar}{i\Phi''(x_0)}}

$$ (eq-spa-result)

:::

#### Conditions for Validity

The SPA is valid when:

1. **Clear stationary point**: There is a unique point (or isolated points) where $d\Phi/dx = 0$.
2. **Large second derivative**: $|\Phi''(x_0)|$ is not too small, so the quadratic expansion is accurate.
3. **Slowly-varying amplitude**: $f(x)$ doesn't oscillate rapidly; it varies over a scale much longer than the stationary region.
4. **Small $\hbar$**: Equivalently, $\Phi \gg \hbar$ (the classical action is large compared to $\hbar$).

#### Region of Significant Contribution

The width of the stationary region is estimated by the condition that the phase varies by $\sim 1$ radian:

$$

\Phi''(x - x_0)^2 \sim \hbar \quad \Rightarrow \quad (x - x_0) \sim \sqrt{\hbar/\Phi''}

$$ (eq-stationary-width)

**Small $\hbar$** means this region is **narrow**: quantum fluctuations are confined to a region of size $\sim \sqrt{\hbar}$, while the amplitude $f$ varies over distance $\sim 1$. Thus $f$ is nearly constant over the stationary region.

### Why the Classical Path Dominates

In the path integral,

$$

K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \int_{\text{all paths}} e^{iS[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$ (eq-path-integral-reminder)

the **phase** is $\Phi[\boldsymbol{x}] = S[\boldsymbol{x}]/\hbar$, where $S$ is the **classical action**.

#### Classical Path Satisfies Stationarity

The **classical path** $\boldsymbol{x}_{\text{cl}}(t)$ is defined by the principle of stationary action:

$$

\frac{\delta S}{\delta \boldsymbol{x}(t)}\bigg|_{\boldsymbol{x}_{\text{cl}}} = 0

$$ (eq-euler-lagrange-functional)

This is the **Euler-Lagrange equation**—Newton's equations of motion in Lagrangian form.

By the stationary phase approximation, paths with $\delta S = 0$ (the classical path and a small neighborhood around it) **interfere constructively**—their phases are nearly identical, so amplitudes add coherently.

#### Destructive Interference Away from the Classical Path

Consider two neighboring paths $\boldsymbol{x}_1(t)$ and $\boldsymbol{x}_2(t) = \boldsymbol{x}_1(t) + \delta\boldsymbol{x}(t)$ that differ slightly. If neither is the classical path, their actions differ:

$$

\Delta S = S[\boldsymbol{x}_2] - S[\boldsymbol{x}_1] = \int \frac{\delta S}{\delta \boldsymbol{x}} \delta\boldsymbol{x} \, dt + O((\delta\boldsymbol{x})^2)

$$

Since $\delta S/\delta\boldsymbol{x} \ne 0$ away from the classical path, infinitesimal variations $\delta\boldsymbol{x}$ produce changes in action $\Delta S$ that are **not second-order small**. Thus:

- Paths clustered around $\boldsymbol{x}_{\text{cl}}$ have action differences $\sim \sqrt{\hbar}$ (or smaller), giving phase differences $\sim 1$—they **interfere coherently**.
- Paths at distance $\sim 1$ from $\boldsymbol{x}_{\text{cl}}$ have action differences $\sim 1$ (in units of $\hbar$), giving phase differences $\sim 1/\hbar \gg 1$—they **oscillate rapidly and cancel**.

:::{admonition} Quantum-to-classical transition
:class: note

In the **classical limit** $\hbar \to 0$:
- The coherent region shrinks to zero size.
- Only the single classical path remains.
- All other paths are suppressed by rapid oscillation.
- The particle follows a definite trajectory.

In the **quantum regime** (small, complex systems):
- Many paths have action $\sim \hbar$ and contribute significantly.
- Quantum superposition of all paths gives interference patterns.
- The particle has no definite trajectory; it explores all possibilities with different probabilities.
:::

### The Second Variation and Quantum Fluctuations

The path integral is dominated by configurations close to the classical path. Expanding the action to second order gives:

$$

S[\boldsymbol{x}_{\text{cl}} + \delta\boldsymbol{x}] \approx S[\boldsymbol{x}_{\text{cl}}] + \frac{1}{2}\int dt \, \frac{\delta^2 S}{\delta\boldsymbol{x}^2} (\delta\boldsymbol{x})^2

$$

The second-order term is a **Gaussian functional integral** that can be evaluated analytically, yielding the propagator in the **stationary phase approximation**:

$$

K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) \approx e^{iS_{\text{cl}}/\hbar} \sqrt{\frac{1}{\det(\delta^2 S/\delta\boldsymbol{x}^2)}}

$$

This includes both the classical action and a **quantum correction factor** from fluctuations.

:::{admonition} Derivation: Second Variation and Quantum Corrections
:class: dropdown note

To understand quantum corrections beyond the classical path, expand the action to second order around the classical path:

$$

S[\boldsymbol{x}_{\text{cl}} + \delta\boldsymbol{x}] = S[\boldsymbol{x}_{\text{cl}}] + \int \frac{\delta^2 S}{\delta\boldsymbol{x}^2} (\delta\boldsymbol{x})^2 \, dt + \cdots

$$ (eq-action-second-variation)

The **second variation** $\delta^2 S/\delta\boldsymbol{x}^2$ is a differential operator (functional Hessian). For a particle in a potential $V(\boldsymbol{x})$:

$$

\frac{\delta^2 S}{\delta\boldsymbol{x}(t)^2} = -m\frac{d^2}{dt^2} - V''(\boldsymbol{x}_{\text{cl}})

$$ (eq-second-variation-explicit)

The path integral over fluctuations $\delta\boldsymbol{x}$ becomes a **Gaussian functional integral**, which can be evaluated using the functional generalization of Gaussian integrals (related to the **functional determinant** or **path integral measure**).

**Result**: 

$$

K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) \approx e^{iS_{\text{cl}}/\hbar} \sqrt{\frac{1}{\det(\delta^2 S/\delta\boldsymbol{x}^2)}} \times \text{(quantum correction factor)}

$$ (eq-path-integral-spa)

The **phase** is determined by the classical action; the **amplitude correction** comes from quantum fluctuations. This is the essence of the **semiclassical approximation**.

:::



:::{example} Example: Particle in a Box
:class: example

For a free particle in a box (no potential), the classical action for a straight-line path from $\boldsymbol{x}_i$ to $\boldsymbol{x}_f$ in time $T$ is

$$

S_{\text{cl}} = \frac{m(\boldsymbol{x}_f - \boldsymbol{x}_i)^2}{2T}

$$ (eq-free-action-cl)

For an electron moving 1 cm in 1 second:

$$

S_{\text{cl}} \sim (10^{-30} \text{ kg})(0.01 \text{ m})^2/(1 \text{ s}) \sim 10^{-34} \text{ J·s}

$$

Compared to $\hbar \sim 10^{-34}$ J·s, we have $S_{\text{cl}} \sim \hbar$—quantum effects are important.

For a baseball (1 kg) moving 1 m in 1 second:

$$

S_{\text{cl}} \sim (1)(1)^2/(1) \sim 1 \text{ J·s} = 10^{34} \hbar

$$

The classical action is enormous—the path integral is totally dominated by the single classical path. Quantum effects vanish. The ball follows Newton's laws.
:::
### The Correspondence Principle

The stationary phase analysis quantifies the **correspondence principle**: Quantum mechanics reduces to classical mechanics in the limit where $S \gg \hbar$ (i.e., macroscopic systems).

| Regime | $S/\hbar$ | Behavior |
|--------|----------|----------|
| Classical | $\gg 1$ | Single path dominates; classical trajectory |
| Semiclassical | $\sim 1$ | Classical path + quantum fluctuations |
| Quantum | $\lesssim 1$ | All paths contribute; wave-like behavior |

### Summary

- **Stationary phase approximation**: Integrals with rapidly oscillating phase are dominated by points where the phase is stationary.
- **Classical path selection**: The path integral selects the classical path (where $\delta S = 0$) through constructive interference.
- **Destructive interference**: Paths far from the classical trajectory have rapidly varying action—their phases cancel.
- **Quantum fluctuations**: Fluctuations around the classical path have action $\sim \hbar$ and contribute quantum corrections.
- **Classical limit**: As $\hbar \to 0$ or $S \gg \hbar$, the coherent region shrinks and only the classical path survives.
- **Correspondence principle**: Quantum mechanics has classical mechanics as a limit; the transition is controlled by the ratio $S/\hbar$.
