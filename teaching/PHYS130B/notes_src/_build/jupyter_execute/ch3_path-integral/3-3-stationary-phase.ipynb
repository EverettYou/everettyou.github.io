# 3.3 Stationary Phase

### Overview

The path integral sums over all possible paths, each weighted by a phase $e^{iS/\hbar}$. Yet in everyday life, particles follow definite classical trajectories, not all paths. The **stationary phase approximation** explains this: when $\hbar$ is small, most paths cancel through destructive interference, and only paths near the classical trajectory survive. This section develops the semiclassical limit of quantum mechanics through the stationary phase approximation and introduces the **WKB method**, a powerful tool for approximate solutions when quantum corrections to classical behavior are small.

**Why it matters**: The path integral provides a bridge between the classical principle of stationary action and quantum mechanics. The stationary phase approximation makes this bridge explicit: it shows how classical mechanics emerges from quantum mechanics as the phase becomes sharply peaked around the classical path.
