# 3.2 Propagator

### Overview

The **propagator** $K$ is the quantum kernel that evolves a wavefunction forward in time. In the path integral framework, the propagator is the **sum over all paths** between two spacetime points, each weighted by its phase. This section develops the propagator from the path integral, introduces time-slicing to make the functional integral concrete, and evaluates the free-particle case exactly.
