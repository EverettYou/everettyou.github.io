# 3.4 Imaginary Time

### Overview

The path integral was developed as a tool for computing quantum dynamics (real time evolution). Yet a remarkable **analytic continuation** to imaginary time reveals a deep connection between quantum mechanics and classical statistical mechanics. By replacing time $t$ with imaginary time $\tau = -it$, the path integral becomes a sum weighted by the Boltzmann factor, connecting the quantum propagator to thermal averages and the partition function. This section explores this powerful duality.
