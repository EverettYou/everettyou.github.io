## Lecture Notes

The Wick-rotated path integral naturally connects quantum mechanics to statistical mechanics: the partition function and thermal density matrix emerge as path integrals in imaginary time. This reveals a deep unity—the Boltzmann distribution $e^{-\beta E}$ is the imaginary-time evolution operator, making temperature and inverse time dual concepts.

### The Density MatrixIn quantum statistical mechanics, the **density matrix** at temperature $T$ is

$$\rho(T) = \frac{e^{-\beta H}}{Z(\beta)}

$$ (eq-density-matrix)

where:- $H$ is the Hamiltonian.- $\beta = 1/(k_B T)$ is the **inverse temperature**.- $Z(\beta) = \text{Tr}(e^{-\beta H})$ is the partition function (normalization).The density matrix is **Hermitian** ($\rho^\dagger = \rho$) and normalized ($\text{Tr}(\rho) = 1$). It encodes all statistical information about the system at thermal equilibrium.

#### Thermal AverageThe **thermal average** (expectation value at temperature $T$) of an observable $\hat{O}$ is

$$\langle O \rangle_T = \text{Tr}(\rho \hat{O}) = \frac{\text{Tr}(e^{-\beta H} \hat{O})}{\text{Tr}(e^{-\beta H})}

$$ (eq-thermal-average)

### The Imaginary-Time Propagator as Density MatrixThe **imaginary-time propagator** over a time interval $\beta\hbar$ is

$$K_E(\boldsymbol{x}', \beta\hbar; \boldsymbol{x}) = \langle \boldsymbol{x}' | e^{-\beta H} | \boldsymbol{x} \rangle

$$ (eq-imaginary-prop-density)

This is precisely the **density matrix in position representation**. The path integral over imaginary time from $\tau = 0$ to $\tau = \beta\hbar$ gives:

$$K_E(\boldsymbol{x}', \beta\hbar; \boldsymbol{x}) = \int_{\boldsymbol{x}(0)=\boldsymbol{x}}^{\boldsymbol{x}(\beta\hbar)=\boldsymbol{x}'} e^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$ (eq-density-path-integral)

**Diagonal elements** ($\boldsymbol{x}' = \boldsymbol{x}$):

$$K_E(\boldsymbol{x}, \beta\hbar; \boldsymbol{x}) = \langle \boldsymbol{x} | e^{-\beta H} | \boldsymbol{x} \rangle

$$The **local density of states** can be extracted from this diagonal element.

### Partition Function from Path Integral:::{admonition} Derivation: Partition Function from Imaginary-Time Path Integral
:class: dropdown note

The partition function is the **trace** of the density matrix:

$$Z(\beta) = \text{Tr}(e^{-\beta H}) = \int d^3\boldsymbol{x} \, K_E(\boldsymbol{x}, \beta\hbar; \boldsymbol{x})

$$ (eq-partition-trace-path)

Inserting the path integral:

$$Z(\beta) = \int d^3\boldsymbol{x} \int_{\text{periodic in } \tau} e^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$The **periodicity condition** $\boldsymbol{x}(\beta\hbar) = \boldsymbol{x}(0)$ arises naturally from the trace (i.e., the integration over initial positions).

:::



### Thermal Averages from Path IntegralTo compute a thermal average $\langle O \rangle = \text{Tr}(\rho \hat{O})$:

$$\langle O(\boldsymbol{x}) \rangle_T = \frac{1}{Z(\beta)} \int_{\text{periodic}} O(\boldsymbol{x}(\tau_0)) e^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$ (eq-thermal-average-path)

where $\tau_0$ is an arbitrary time on the circle (thermal average is independent of the insertion point).**Key insight**: To measure an observable at temperature $T$, insert it into the path integral at any imaginary time $\tau_0 \in [0, \beta\hbar]$ and integrate over all periodic paths.

### Free Energy and Thermodynamic Quantities

#### Helmholtz Free EnergyThe **Helmholtz free energy** is related to the partition function by

$$F(T, V, N) = -k_B T \ln Z(\beta)

$$ (eq-helmholtz-free-energy)

The path integral provides a direct route to $F$: compute $Z$ via the path integral, then take the logarithm.

#### Entropy

$$S = -\left(\frac{\partial F}{\partial T}\right)_{V,N} = k_B(\ln Z + \beta \langle H \rangle)

$$ (eq-entropy)

#### Internal Energy

$$\langle H \rangle = -\frac{\partial \ln Z}{\partial \beta}

$$ (eq-internal-energy)

All thermodynamic quantities follow from the single quantity $Z(\beta)$, which the path integral computes.

### Quantum-Classical Correspondence at Finite Temperature

#### High-Temperature Limit ($\beta \to 0$)At high temperature, the density matrix becomes sharply peaked near the position of minimum potential energy:

$$\rho(\boldsymbol{x}', \boldsymbol{x}) \approx \sqrt{\frac{m}{2\pi\hbar^2 \beta}} e^{-V(\boldsymbol{x})/(k_B T)} \delta(\boldsymbol{x}' - \boldsymbol{x})

$$The system is **classical**: particles bounce around in the potential, governed by Boltzmann statistics $e^{-V/(k_B T)}$.

#### Low-Temperature Limit ($\beta \to \infty$)At zero temperature, the density matrix projects onto the ground state:

$$\rho(\boldsymbol{x}', \boldsymbol{x}) \to |\phi_0(\boldsymbol{x}')\rangle\langle\phi_0(\boldsymbol{x})|

$$where $|\phi_0\rangle$ is the ground state. The system is **purely quantum**: in its lowest energy state, exhibiting zero-point energy and quantum correlations.

### Computational Advantage: Monte Carlo SimulationsThe Euclidean path integral $\int e^{-S_E} \mathcal{D}[\boldsymbol{x}]$ has a **positive weight** (measure), making it suitable for **Monte Carlo importance sampling**:1. Sample configurations $\boldsymbol{x}(\tau)$ with probability $\propto e^{-S_E[\boldsymbol{x}]/\hbar}$.2. Compute observables $O(\boldsymbol{x})$ on each sample.3. Average: $\langle O \rangle \approx \frac{1}{N_\text{samples}} \sum_i O(\boldsymbol{x}_i)$.In contrast, the real-time path integral $\int e^{iS} \mathcal{D}[\boldsymbol{x}]$ has rapidly oscillating phase (sign problem), making it difficult to simulate directly.

:::{admonition} Practical importance:class: noteLattice QCD (quantum chromodynamics) simulations rely entirely on the imaginary-time formulation. Computing thermal properties, equation of state, and phase transitions at finite temperature requires the Euclidean path integral and Monte Carlo. This is how physicists study the quark-gluon plasma and early-universe physics.
:::
:::{admonition} Classical-quantum correspondence:class: seealsoThe transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the correspondence principle.
:::
### Summary
- **Density matrix**: $\rho = e^{-\beta H}/Z$ governs thermal equilibrium; fully encoded by imaginary-time path integral.
- **Imaginary-time propagator**: $K_E(\boldsymbol{x}', \beta\hbar; \boldsymbol{x})$ is the density matrix in position representation.
- **Periodic paths**: Trace condition $\boldsymbol{x}(\beta\hbar) = \boldsymbol{x}(0)$ emerges from summation over states.
- **Thermal averages**: Observables inserted at arbitrary $\tau_0$ in the periodic path integral give $\langle O \rangle_T$.
- **Thermodynamic functions**: Free energy, entropy, internal energy all follow from partition function $Z(\beta)$.
- **Classical limit**: High $T$ ($\beta \to 0$) gives classical Boltzmann distribution.
- **Quantum limit**: Low $T$ ($\beta \to \infty$) projects to ground state.
- **Numerical simulation**: Positive weight $e^{-S_E}$ enables efficient Monte Carlo sampling, crucial for lattice field theory.
