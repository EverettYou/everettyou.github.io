## Lecture Notes

:::{admonition} Definition
:class: important

The **Wick rotation** from real time $t$ to imaginary time $\tau = it/\hbar$ transforms the oscillatory quantum path integral into an exponentially damped sum. This analytic continuation reveals that quantum mechanics and thermal statistical mechanics are intimately related—they are two faces of the same mathematical structure.
:::

### The Wick Rotation

In quantum field theory and statistical mechanics, analytic continuation from real time $t$ to **imaginary time** $\tau = -it/\hbar$ simplifies calculations.

#### Definition

The **Wick rotation** is the transformation

$$t \to -\mathrm{i}\tau, \quad \text{where} \quad \tau \in \mathbb{R}$$


Equivalently, $\tau = it/\hbar$ is the **imaginary time**, with dimension of time. If thermal context applies, write $\tau \in [0, \beta\hbar]$ where $\beta = 1/(k_B T)$ is the **inverse temperature**.

#### Effect on the Action

:::{admonition} Derivation: Wick Rotation Transform
:class: dropdown note

The **real-time action** is

$$S[\boldsymbol{x}] = \int_0^T \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})\right) dt$$


Under $t \to -\mathrm{i}\tau$, we have $dt \to -i d\tau$ and $\dot{\boldsymbol{x}} = d\boldsymbol{x}/dt \to i d\boldsymbol{x}/d\tau$. Thus:

$$S[\boldsymbol{x}] \to \int_0^{\beta\hbar} \left(\frac{m}{2}(i\dot{\boldsymbol{x}})^2 - V(\boldsymbol{x})\right) (-i) d\tau = -i \int_0^{\beta\hbar} \left(-\frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})\right) d\tau$$


The minus signs combine to give:

$$S[\boldsymbol{x}] \to -\int_0^{\beta\hbar} \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 + V(\boldsymbol{x})\right) d\tau$$


Defining the **Euclidean action** as

$$S_E[\boldsymbol{x}] = \int_0^{\beta\hbar} \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 + V(\boldsymbol{x})\right) d\tau$$


we have $S \to -iS_E$, so

$$\mathrm{e}^{\mathrm{i}S/\hbar} \to \mathrm{e}^{-S_E/\hbar}$$


:::

The **rapidly oscillating phase** $\mathrm{e}^{\mathrm{i}S/\hbar}$ becomes a **rapidly decaying exponential** $\mathrm{e}^{-S_E/\hbar}$. This convergence is crucial for numerical simulations and analytical approximations.

:::{admonition} Key Observation: Why Wick Rotation Matters
:class: important

Real-time path integrals with $\mathrm{e}^{\mathrm{i}S/\hbar}$ are **highly oscillatory**—contributions from different paths interfere constructively and destructively, making direct integration intractable. 

After Wick rotation, the Euclidean path integral with $\mathrm{e}^{-S_E/\hbar}$ is **exponentially damped**—contributions from paths with large action are exponentially suppressed, enabling:
- Numerical Monte Carlo simulations
- Perturbative expansions
- Semi-classical approximations
:::

### Euclidean Path Integral

The **Euclidean path integral** is

$$Z(\beta) = \int_{\text{periodic}} \mathrm{e}^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]$$


where the integral is over paths that are **periodic in imaginary time**: $\boldsymbol{x}(\tau + \beta\hbar) = \boldsymbol{x}(\tau)$.

**Key properties**:

1. **Convergence**: The exponential $\mathrm{e}^{-S_E/\hbar}$ decays exponentially for large actions. Paths with large Euclidean action are suppressed. This makes numerical simulation feasible.

2. **Positive measure**: Unlike the oscillating real-time path integral, the Euclidean integral has a well-defined probability measure (positive definite). This is essential for Monte Carlo simulations.

3. **Periodicity**: The periodicity in imaginary time $\tau$ reflects the **cyclic nature** of the thermal density matrix in quantum statistical mechanics.

### Connection to the Partition Function

In quantum statistical mechanics, the **partition function** at temperature $T$ is

$$Z(T) = \text{Tr}(\mathrm{e}^{-\beta H}), \quad \beta = \frac{1}{k_B T}$$


where $H$ is the Hamiltonian and the trace sums over all quantum states.

:::{admonition} Path Integral Representation of Partition Function
:class: important

$$Z(\beta) = \int_{\text{periodic}} \mathrm{e}^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]$$


where the sum is over all paths periodic in imaginary time with period $\beta\hbar = 1/(k_B T)$.

The partition function $Z = \text{Tr}(\mathrm{e}^{-\beta H})$ equals the functional integral over periodic paths in Euclidean time. This is a profound connection:
- **Trace**: sums over all eigenstates
- **Path integral**: sums over all closed paths in imaginary time
:::

### Analytic Continuation

Results computed in imaginary time (Euclidean path integral) can be analytically continued back to real time through various methods.

#### Direct Continuation

If a result is known as a function $F(\tau)$ on the imaginary-time axis, it can often be extended to real time $t$ by replacing $\tau \to it$:

$$F(t) = F(\tau)|_{\tau \to it}$$


This works if $F$ is analytic in the complex-$\tau$ plane.

#### Matsubara Frequency Expansion

For correlators or Green's functions, expand in **Matsubara frequencies**:

$$G(\tau) = \sum_{n=-\infty}^{\infty} \frac{1}{\beta\hbar} \tilde{G}(\mathrm{i}\omega_n) \mathrm{e}^{-\mathrm{i}\omega_n \tau}$$


where $\omega_n = 2\pi n/(\beta\hbar)$ are the **Matsubara frequencies** (quantized imaginary frequencies). Continuing to real frequency $\omega$ gives the real-time result.

### Physical Interpretation

:::{admonition} Discussion: Imaginary-Time Path Integrals and Thermal States
:class: dropdown warning

Wick rotation replaces $t \to -\mathrm{i}\tau$ (imaginary time) to convert the quantum path integral into a statistical mechanics path integral. This is a powerful mathematical trick, but is it more than that?

- If we compute physics in imaginary time and rotate back to real time, are we computing the *same* physics? Can all real-time amplitudes be extracted from imaginary-time calculations?
- In imaginary time, the path integral weight is $\mathrm{e}^{-S_E/\hbar}$ where $S_E$ is the Euclidean action. This looks like Boltzmann weight $\mathrm{e}^{-\beta E}$ (with identification $\beta = 1/k_B T$). Is this resemblance accidental, or does it reveal deep structure?
- For a quantum field at finite temperature, the path integral is over periodic paths in imaginary time (period $\beta = 1/(k_B T)$). Can you extend this connection beyond field theory to quantum mechanics?
- Some states (like instantons) exist in imaginary time but have no real-time counterpart. Does this mean they don't physically exist, or is imaginary time equally "real" as real time?
:::
### Temperature Limits

#### Low Temperature Limit

As $T \to 0$, we have $\beta \to \infty$. The period of the imaginary-time path becomes infinite. Only the **ground state** (lowest energy) contributes significantly to the path integral:

$$Z(T \to 0) \approx \mathrm{e}^{-\beta E_0}$$


where $E_0$ is the ground-state energy. Thus, **zero-temperature path integrals govern ground-state properties**.

#### High Temperature Limit

As $T \to \infty$, $\beta \to 0$. Many paths contribute, and the quantum kinetic energy term $\propto \dot{\boldsymbol{x}}^2$ becomes negligible. The partition function approaches the **classical statistical mechanics result**:

$$Z(T \to \infty) \approx \int \mathrm{e}^{-V(\boldsymbol{x})/(k_B T)} d^3\boldsymbol{x}$$


This is the **classical partition function**, weighted only by Boltzmann factors for potential energy. Quantum effects vanish at high temperature.

:::{admonition} Note: Classical-Quantum Correspondence
:class: note

The transition from quantum to classical mechanics is smooth. 
- At **low temperatures** ($\beta E_0 \gg 1$): quantum effects dominate; ground state governs all properties
- At **intermediate temperatures**: both ground and excited states contribute; quantum-classical crossover
- At **high temperatures** ($k_B T \gg \hbar\omega$): classical statistical mechanics dominates; kinetic energy negligible

The Wick rotation and partition function naturally interpolate between these regimes.
:::

### Summary

- **Wick rotation**: $t \to -\mathrm{i}\tau$ transforms real-time oscillating path integral to imaginary-time converging one
- **Euclidean action**: $S_E = \int (\frac{m}{2}\dot{\boldsymbol{x}}^2 + V) d\tau$; kinetic term positive-definite
- **Phase to exponential**: $\mathrm{e}^{\mathrm{i}S/\hbar} \to \mathrm{e}^{-S_E/\hbar}$ makes path integral numerically tractable
- **Partition function**: Euclidean path integral over periodic paths equals $Z = \text{Tr}(\mathrm{e}^{-\beta H})$
- **Periodicity**: Imaginary-time period $\beta\hbar = 1/(k_B T)$ encodes temperature
- **Analytic continuation**: Results from imaginary time can be continued to real time via analytic functions or Matsubara frequencies
- **Temperature limits**: $T \to 0$ selects ground state; $T \to \infty$ recovers classical mechanics
- **Statistical-quantum duality**: Wick rotation reveals thermal statistical mechanics as the imaginary-time version of quantum mechanics
