## Lecture Notes

### The Adiabatic Theorem and Phase Evolution

When a quantum system's Hamiltonian $\hat{H}(\boldsymbol{R}(t))$ changes slowly compared to the gap $\Delta E$ between eigenlevels, the system remains in its instantaneous eigenstate:

$$

\hat{H}(\boldsymbol{R})|n(\boldsymbol{R})\rangle = E_n(\boldsymbol{R})|n(\boldsymbol{R})\rangle

$$

This is the **adiabatic theorem**. If the system starts in eigenstate $|n(\boldsymbol{R}_0)\rangle$, it will stay in $|n(\boldsymbol{R}(t))\rangle$ throughout evolution.

However, the quantum state acquired during this process is not merely:

$$

|\psi(t)\rangle = \mathrm{e}^{-\mathrm{i}E_n(t)/\hbar} |n(\boldsymbol{R}(t))\rangle

$$

This expression misses a **geometric phase** — Berry's phase — that depends on the path traced through parameter space, not just the initial and final states.

:::{admonition} The Phase Evolution
:class: important

The full adiabatic evolution of the $n$-th eigenstate is:

$$

|\psi(t)\rangle = \mathrm{e}^{\mathrm{i}\gamma_n(t)} \mathrm{e}^{-\mathrm{i}\theta_n(t)} |n(\boldsymbol{R}(t))\rangle

$$

where:
- **Dynamical phase**: $\theta_n(t) = \frac{1}{\hbar}\int_0^t E_n(\boldsymbol{R}(t'))\,\mathrm{d}t'$
- **Geometric phase** (Berry phase): $\gamma_n(t) = \mathrm{i}\int_0^t \langle n(\boldsymbol{R})|{\boldsymbol{\nabla}_{\boldsymbol{R}}|n(\boldsymbol{R})\rangle} \cdot \dot{\boldsymbol{R}}\,\mathrm{d}t'$

:::

### Berry Connection and Berry Curvature

Define the **Berry connection**:

$$

\mathcal{A}_n(\boldsymbol{R}) = \mathrm{i}\langle n(\boldsymbol{R})|\nabla_{\boldsymbol{R}}|n(\boldsymbol{R})\rangle

$$

It is a vector field in parameter space. Then:

$$

\gamma_n(t) = \int_C \mathcal{A}_n(\boldsymbol{R}) \cdot \mathrm{d}\boldsymbol{R}

$$

where $C$ is the path in parameter space from initial to final $\boldsymbol{R}(t)$.

:::{admonition} Berry Connection as Geometric Vector Potential
:class: important

The **Berry connection** $\mathcal{A}_n(\boldsymbol{R}) = \mathrm{i}\langle n(\boldsymbol{R})|\nabla_{\boldsymbol{R}}|n(\boldsymbol{R})\rangle$ plays the role of a vector potential in parameter space, analogous to the electromagnetic vector potential $\boldsymbol{A}(\boldsymbol{r})$.

Key properties:
- **Gauge-dependent**: Transforms as $\mathcal{A}_n \to \mathcal{A}_n + \nabla_{\boldsymbol{R}}\xi$ under phase rotation $|n\rangle \to \mathrm{e}^{\mathrm{i}\xi}|n\rangle$
- **Line integral gives Berry phase**: $\gamma_n = \int_C \mathcal{A}_n \cdot \mathrm{d}\boldsymbol{R}$ (for open paths, gauge-dependent)
- **Real-valued**: $\mathcal{A}_n$ is always real because $\langle n|n\rangle = 1 \Rightarrow \langle n|\nabla|n\rangle + \text{c.c.} = 0$

:::

The **Berry curvature** is:

$$

\mathcal{F}_n(\boldsymbol{R}) = \nabla_{\boldsymbol{R}} \times \mathcal{A}_n(\boldsymbol{R})

$$

By Stokes' theorem, for a closed path enclosing a surface $S$ in parameter space:

$$

\gamma_n = \oint_C \mathcal{A}_n \cdot \mathrm{d}\boldsymbol{R} = \int_S \mathcal{F}_n \cdot \mathrm{d}\boldsymbol{S}

$$

:::{admonition} Berry Curvature and Chern Number
:class: important

The **Berry curvature** $\mathcal{F}_{n,\mu\nu}(\boldsymbol{R}) = \partial_{R_\mu}\mathcal{A}_{n,\nu} - \partial_{R_\nu}\mathcal{A}_{n,\mu}$ is a rank-2 tensor (the "field strength" of the Berry connection).

Arises from off-diagonal couplings:

$$

\mathcal{F}_{n,\mu\nu} = -2\,\mathrm{Im}\sum_{m\neq n} \frac{\langle n|\partial_\mu \hat{H}|m\rangle\langle m|\partial_\nu \hat{H}|n\rangle}{(E_m - E_n)^2}

$$

For a 2D parameter space (e.g., crystal momentum $\boldsymbol{k}$), the **Chern number** is the topological integral:

$$

C_n = \frac{1}{2\pi}\int \mathcal{F}_n \, \mathrm{d}A \in \mathbb{Z}

$$

This is always an integer—a topological invariant that cannot change without closing the band gap.

:::

:::{admonition} Derivation: Berry Curvature Formula
:class: dropdown

The Berry curvature can be expressed in terms of the Hamiltonian. Acting on eigenstate $|n(\boldsymbol{R})\rangle$:

$$

\partial_\mu \mathcal{A}_{n,\nu} = \mathrm{i}\partial_\mu\langle n|\partial_\nu n\rangle = \mathrm{i}[\partial_\mu\langle n|]\partial_\nu|n\rangle + \mathrm{i}\langle n|\partial_{\mu}\partial_\nu|n\rangle

$$

The off-diagonal component comes from:

$$

\mathrm{i}\partial_\mu\langle n| = \mathrm{i}\sum_{m\neq n}\frac{\langle m|\partial_\mu\hat{H}|n\rangle}{(E_n - E_m)}\langle m|

$$

Combining and subtracting to form the curl gives the result above.

:::

### Gauge Invariance and Physical Observables

The geometric phase is *physically* well-defined only for closed paths. This is because we can perform a **gauge transformation**:

$$

|n(\boldsymbol{R})\rangle \to \mathrm{e}^{\mathrm{i}\phi(\boldsymbol{R})}|n(\boldsymbol{R})\rangle

$$

Under this redefinition, the Berry connection transforms as:

$$

\mathcal{A}_n \to \mathcal{A}_n + \nabla_{\boldsymbol{R}}\phi

$$

For an open path, the Berry phase changes. But for a **closed path**:

$$

\oint_C (\mathcal{A}_n + \nabla_{\boldsymbol{R}}\phi) \cdot \mathrm{d}\boldsymbol{R} = \oint_C \mathcal{A}_n \cdot \mathrm{d}\boldsymbol{R} + \oint_C \nabla\phi \cdot \mathrm{d}\boldsymbol{R} = \oint_C \mathcal{A}_n \cdot \mathrm{d}\boldsymbol{R}

$$

The gradient term vanishes around a closed loop, so **$\gamma_n$ is gauge-invariant for closed paths** (modulo $2\pi$).

:::{admonition} Gauge Invariance: Connection vs. Curvature
:class: important

- **Berry connection** $\mathcal{A}_n(\boldsymbol{R})$: **Gauge-dependent** (transforms like vector potential)
- **Berry curvature** $\mathcal{F}_n(\boldsymbol{R})$: **Gauge-invariant** (transforms like magnetic field)
- **Berry phase on closed loops**: **Gauge-invariant** (depends only on geometry, not phase choice)
- **Berry phase on open paths**: **Gauge-dependent** (depends on choice of phase convention)

This mirrors electromagnetism: $\boldsymbol{A}$ is gauge-dependent, but $\boldsymbol{B} = \nabla \times \boldsymbol{A}$ and $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{\ell}$ around closed loops are gauge-invariant.

:::

### Spin-1/2 in a Rotating Magnetic Field

**Example**: A spin-1/2 particle in a slowly rotating magnetic field $\boldsymbol{B}(t)$ with fixed magnitude $B_0$ but rotating direction.

For adiabatic evolution, the spin stays aligned with the instantaneous field. As $\boldsymbol{B}$ traces a closed path on the unit sphere, sweeping solid angle $\Omega$, the Berry phase is:

$$

\gamma = -\frac{\Omega}{2}

$$

The **negative sign** reflects the spinor's intrinsic properties. For a path that encloses a full hemisphere ($\Omega = 2\pi$), $\gamma = -\pi$.

This is intimately connected to the fact that a spin-1/2 wave function has $4\pi$ periodicity (a $2\pi$ rotation about any axis gives $-1$, not $+1$). The Berry phase partially compensates this geometric structure.

:::{admonition} Solid Angle Formula for Spin-1/2
:class: important

For a spin-1/2 in a rotating magnetic field, as the field direction traces a closed path enclosing solid angle $\Omega$ on the unit sphere:

$$

\gamma_{\text{Berry}} = -\frac{\Omega}{2}

$$

This is called the **Pancharatnam-Berry phase** or **geometric phase**. The factor $-1/2$ is characteristic of spin-1/2; higher spins have different coefficients.

The solid angle $\Omega$ can be computed from the path using spherical geometry. For a cone of half-angle $\theta_0$ traversed once: $\Omega = 2\pi(1 - \cos\theta_0)$.

:::

:::{admonition} Derivation: Solid Angle and Berry Phase
:class: dropdown

For a spin-1/2 in field $\boldsymbol{B}(t) = B_0(\sin\theta(t)\cos\phi(t), \sin\theta(t)\sin\phi(t), \cos\theta(t))$, the instantaneous ground state is:

$$

|+(\theta,\phi)\rangle = \begin{pmatrix} \cos(\theta/2) \\ \sin(\theta/2)\mathrm{e}^{\mathrm{i}\phi} \end{pmatrix}

$$

The Berry connection is:

$$

\mathcal{A} = \mathrm{i}\langle +|\nabla|+\rangle

$$

Computing $\langle +|\partial_\theta |+\rangle$ and $\langle +|\partial_\phi|+\rangle$:

$$

\langle +|\partial_\theta|+\rangle = 0, \quad \langle +|\partial_\phi|+\rangle = -\frac{1}{2}

$$

Thus $\mathcal{A} = -\frac{1}{2}\mathrm{d}\phi$, and:

$$

\gamma = \oint \mathcal{A} \cdot \mathrm{d}\boldsymbol{R} = -\frac{1}{2}\oint \mathrm{d}\phi = -\frac{\Omega}{2}

$$

where $\Omega = \int \mathrm{d}\Omega$ is the solid angle.

:::

### Physical Interpretation: Holonomy

Berry phase is a **holonomy** — a phase acquired by parallel transport of a quantum state around a closed path in an abstract geometric space (the space of Hamiltonian parameters).

In general relativity, parallel transport around a closed loop in curved spacetime results in a rotation of the vector. Similarly, the quantum state's phase is "rotated" by geometric curvature in the parameter space landscape.

This perspective unifies many quantum phenomena:
- Molecular spectra (Berry phase in electronic coordinates)
- Transport in solids (Berry phase in crystal momentum space)
- Photon polarization (Berry phase in direction of propagation)
