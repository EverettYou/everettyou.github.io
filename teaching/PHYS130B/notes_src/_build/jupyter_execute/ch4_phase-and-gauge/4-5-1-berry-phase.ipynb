## Lecture Notes

As a quantum system evolves adiabatically through parameter space along a closed path, it acquires not just the dynamical phase from energy, but an additional geometric phase—the Berry phase—that depends only on the path's geometry, not on evolution speed. This phase arises from the curvature of the eigenstate space and reveals that quantum mechanics has a hidden geometric structure. Berry phase connects gauge theory, topology, and observable interference phenomena, and appears throughout quantum systems: atoms in slowly-rotating fields, electrons in slowly-varying potentials, and topological condensed matter.
:::{admonition} Berry Phase Survives Adiabatic Approximation
:class: important

The adiabatic approximation ensures no transitions between states. Yet the Berry phase accumulates even when the system never leaves its instantaneous eigenstate. This is the magic of the geometric phase.
:::


### The Adiabatic Theorem

The **adiabatic theorem** is a fundamental principle in quantum mechanics that applies when parameters in the Hamiltonian change slowly.

**Statement**: If a quantum system starts in an eigenstate of $\hat{H}(\boldsymbol{R}(0))$ and the parameters $\boldsymbol{R}(t)$ change slowly (adiabatically), then the system **remains in the instantaneous eigenstate** throughout the evolution:

$$

|\psi(t)\rangle = e^{i\gamma_n(t)}e^{-i\phi_d(t)}|n(\boldsymbol{R}(t))\rangle

$$

The adiabatic condition requires:

$$

\left|\frac{d\boldsymbol{R}/dt}{E_{m} - E_n}\right| \ll 1 \quad (m \neq n)

$$

That is, parameters must change much more slowly than the intrinsic timescale set by energy level spacings.

**Rigorous derivation**: The adiabatic theorem can be derived rigorously from time-dependent perturbation theory. See **[§5.2.3](../ch5_perturbation-theory/5-2-3-applications.ipynb)** for the full derivation showing how adiabatic evolution naturally emerges when the perturbation timescale is much longer than energy gaps.

### Dynamical Phase

The **dynamical phase** is the phase accumulated from the time evolution of an energy eigenstate:

$$

\phi_d(t) = \frac{1}{\hbar}\int_0^t E_n(\boldsymbol{R}(t'))\, dt'

$$

This is exactly what we expect from the Schrödinger equation $i\hbar\partial_t|\psi\rangle = \hat{H}|\psi\rangle$. For a state with fixed energy $E_n$, the phase factor is simply $e^{-iE_n t/\hbar}$.

The dynamical phase:
- Depends on the evolution **time** (explicitly through the integral)
- Depends on the **actual path** taken through parameter space (via $E_n(\boldsymbol{R}(t))$)
- Is easily understood in terms of energy considerations

### Berry Phase: The Geometric Contribution

Beyond the dynamical phase, an adiabatically-evolving eigenstate acquires an additional **geometric phase**, discovered by Michael Berry in 1984:

$$

\gamma_n(C) = i\oint_C \langle n(\boldsymbol{R}) | \nabla_\boldsymbol{R} n(\boldsymbol{R}) \rangle \cdot d\boldsymbol{R}

$$

where the integral is over a closed loop $C$ in parameter space.

**Remarkable properties**:

1. **Path-dependent, not time-dependent**: The Berry phase depends only on the geometric shape of the path $C$, NOT on how long the evolution takes. Fast or slow adiabatic evolution gives the same Berry phase!

2. **Geometric origin**: It depends on the geometry of the eigenstate in Hilbert space as a function of parameters.

3. **Closed paths only**: For open paths, the phase can be gauged away by redefining the phase of the eigenstate. Only for closed loops is the phase well-defined and unambiguous.

4. **Gauge ambiguity**: The Berry phase depends on the phase convention chosen for $|n(\boldsymbol{R})\rangle$. Different phase choices shift the Berry phase by multiples of $2\pi$.

### Berry Connection

Define the **Berry connection** (a gauge field in parameter space):

$$

\boldsymbol{A}_{n}(\boldsymbol{R}) = i\langle n(\boldsymbol{R}) | \nabla_\boldsymbol{R} n(\boldsymbol{R}) \rangle

$$

This is a vector field in parameter space with dimensions of inverse length. It encodes how the phase of the eigenstate changes as we move through parameter space.

The Berry phase can then be written compactly as:

$$

\gamma_n(C) = \oint_C \boldsymbol{A}_{n}(\boldsymbol{R}) \cdot d\boldsymbol{R}

$$

This is the **line integral of the Berry connection**, analogous to the phase acquired by a charged particle traveling through an electromagnetic vector potential.

### Berry Curvature

Just as the electromagnetic field is the curl of the vector potential, we define the **Berry curvature** as a 2-form in parameter space:

$$

\Omega_{n,ij}(\boldsymbol{R}) = \partial_{R_i}A_{n,j} - \partial_{R_j}A_{n,i}

$$

By **Stokes' theorem**, the Berry phase can be expressed as a surface integral of the curvature:

$$

\gamma_n(C) = \oint_C \boldsymbol{A}_n \cdot d\boldsymbol{R} = \int_S \boldsymbol{\Omega}_n(\boldsymbol{R}) \cdot d\boldsymbol{S}

$$

**Key property**: Unlike the Berry connection (which is gauge-dependent), the **Berry curvature is gauge-invariant**. This makes the curvature a fundamental observable quantity that cannot be transformed away by phase redefinitions of the eigenstate.

The Berry curvature:

- Measures the **geometric twist** in the eigenstate Hilbert space
- Encodes observable topological properties

### Explicit Formula for Curvature

The Berry curvature can be expressed in terms of matrix elements of the Hamiltonian derivatives:

$$

\Omega_{n,ij} = -2\text{Im}\sum_{m \neq n}\frac{\langle n | \partial_i \hat{H} | m \rangle \langle m | \partial_j \hat{H} | n \rangle}{(E_n - E_m)^2}

$$

This formula reveals that:
- The curvature involves **off-diagonal matrix elements** of Hamiltonian derivatives
- States with large energy differences contribute weakly (denominator suppression)
- The sum is over all excited states $m 
eq n$
- The formula is intrinsically gauge-invariant (imaginary part of a fraction)

This provides a practical way to calculate Berry curvature from first principles given the Hamiltonian structure.

The Berry curvature can be expressed in terms of other eigenstates:

$$

\Omega_{n,ij} = -2\text{Im}\sum_{m \neq n}\frac{\langle n | \partial_i \hat{H} | m \rangle \langle m | \partial_j \hat{H} | n \rangle}{(E_n - E_m)^2}

$$

This formula shows that the curvature involves matrix elements of the Hamiltonian derivatives weighted by energy differences. States that are far in energy contribute weakly.

:::
:class: example


A paradigmatic example: a spin-1/2 particle in a slowly rotating magnetic field.

Let $\boldsymbol{B}(t)$ rotate around the $z$-axis with angular velocity $\Omega \ll \omega_L$ (much slower than Larmor frequency $\omega_L = g B/(2m_e c)$).

The spin adiabatically follows the magnetic field direction (following the instantaneous eigenstate). As the field completes one full rotation around the $z$-axis, tracing a cone of half-angle $\theta$, the spin acquires Berry phase:

$$

\gamma = \frac{1}{2}\Omega_\text{solid angle} = \frac{1}{2} \cdot 2\pi(1 - \cos\theta) = \pi(1 - \cos\theta)

$$

For a field rotating around the equator ($\theta = \pi/2$):

$$

\gamma = \pi

$$

This is the famous **geometric phase of a spinning particle**.

### Comparison: Geometric vs Dynamical Phase

| Aspect | Dynamical Phase | Geometric Phase |
|---|---|---|
| Formula | $\phi_d = -\int E_n dt/\hbar$ | $\gamma = i\oint \langle n\|\nabla n\rangle \cdot dR$ |
| Dependence | Explicit time integral | Path shape only |
| Speed sensitivity | YES—faster evolution gives smaller phase | NO—independent of speed |
| Closed loops needed | NO—defined for any path | YES—vanishes for open paths |
| Gauge-invariance | $2\pi$ ambiguity | Curvature is gauge-invariant |
| Observable | Direct energy measurement | Requires interference |

### Summary

- **Adiabatic theorem**: Slow parameter change preserves eigenstate (up to phases).
- **Dynamical phase**: Standard energy-dependent phase $\phi_d = -\int E_n dt/\hbar$.
- **Berry phase**: Additional geometric phase $\gamma = i\oint \langle n|\nabla_R n\rangle \cdot dR$ from path geometry.
- **Path-dependent**: Berry phase depends only on path shape, not evolution speed.
- **Berry connection**: Gauge field $\boldsymbol{A}_n = i\langle n|\nabla_R n\rangle$ in parameter space.
- **Berry curvature**: $\boldsymbol{\Omega}_n = \nabla \times \boldsymbol{A}_n$—gauge-invariant field strength.
- **Stokes' theorem**: $\gamma = \oint \boldsymbol{A} \cdot dR = \int \boldsymbol{\Omega} \cdot dS$ connects connection and curvature.
- **Geometric interpretation**: Curvature measures geometric twist of eigenstate in Hilbert space.
- **Closed loops only**: Non-zero Berry phase requires topologically non-trivial loops.
- **Rigorous foundation**: The adiabatic theorem and Berry phase emerge naturally from time-dependent perturbation theory (see **[§5.2.3](../ch5_perturbation-theory/5-2-3-applications.ipynb)**).
