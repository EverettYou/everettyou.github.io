## Lecture Notes

### The Chern Number and Topological Invariants

The **Chern number** is an integer topological invariant that classifies quantum states and energy bands. For a 2D band structure, it is defined as:

$$


C_n = \frac{1}{2\pi}\int_{\text{BZ}} \mathcal{F}_{xy}^{(n)}(\boldsymbol{k})\,\mathrm{d}^2k \in \mathbb{Z}

$$


where the integral is over the Brillouin zone and $\mathcal{F}_{xy}^{(n)}$ is the Berry curvature of band $n$ in momentum space.

The Chern number is called a **topological invariant** because:
1. It is always an integer
2. It cannot change under smooth deformations of the Hamiltonian (unless the band gap closes)
3. It is independent of the choice of gauge (the choice of phase for the Bloch states)

These properties make $C_n$ robust to disorder and perturbations — it characterizes a fundamentally different phase of matter.

:::{admonition} Chern Number Definition
:class: important

The **Chern number** $C_n \in \mathbb{Z}$ of a 2D Bloch band $n$ is a topological invariant:

$$


C_n = \frac{1}{2\pi}\int_{\text{BZ}} \mathcal{F}_{xy}^{(n)}(\boldsymbol{k})\,\mathrm{d}^2k \in \mathbb{Z}

$$


where $\mathcal{F}_{xy}^{(n)}(\boldsymbol{k}) = \partial_{k_x} \mathcal{A}_y^{(n)} - \partial_{k_y} \mathcal{A}_x^{(n)}$ is the **Berry curvature** and $\mathcal{A}^{(n)} = i\langle n(\boldsymbol{k})|\nabla_{\boldsymbol{k}}n(\boldsymbol{k})\rangle$ is the Berry connection.

**Key properties**:
- Always an integer (topological quantization)
- Cannot change under smooth deformations without closing the band gap
- Gauge-invariant: independent of the choice of phase for Bloch states
- Characterizes the winding of the electronic state's phase as momentum traverses the Brillouin zone

:::

### Physical Picture: Berry Flux and Winding

Imagine smoothly deforming a band by continuously changing Hamiltonian parameters. The Berry curvature $\mathcal{F}_{xy}^{(n)}(\boldsymbol{k})$ encodes the "twist" or "winding" of the electron's internal quantum state as we move through momentum space.

Integrating this curvature over the entire Brillouin zone gives the total "magnetic flux" of Berry curvature:

$$


\Phi_{\text{Berry}} = \int_{\text{BZ}} \mathcal{F}_{xy}^{(n)}\,\mathrm{d}^2k = 2\pi C_n

$$


The quantization condition forces $C_n$ to be an integer. You cannot continuously deform the band to reduce this flux by less than one "flux quantum" ($2\pi$) without closing the gap or changing topological structure.

### The TKNN Formula and Integer Quantum Hall Effect

**Theorem** (TKNN, 1982): The quantized Hall conductance of a 2D electron system with filled energy bands is:

$$


\sigma_{xy} = \frac{e^2}{h}\sum_{n=1}^{N_{\text{filled}}} C_n

$$


where the sum runs over all occupied bands and $C_n$ is the Chern number of band $n$.

:::{admonition} TKNN Formula: Quantized Hall Conductance
:class: important

**Thouless, Kohmoto, Nightingale, and Newnham (1982)**: The quantized Hall conductance of a 2D electron system with filled energy bands is:

$$


\sigma_{xy} = \frac{e^2}{h} \sum_{n=1}^{N_{\text{filled}}} C_n

$$


where the sum runs over all occupied bands and $C_n$ is the Chern number of band $n$.

**Physical significance**:
- Relates a macroscopic transport coefficient (Hall conductance) to a microscopic topological invariant (Chern number)
- Hall conductance is quantized in units of $e^2/h$ (no device-dependent factors)
- Quantization is **exact and robust** against disorder, impurities, and perturbations (as long as the band gap remains open)
- This explains the universal plateaus observed in the integer quantum Hall effect

:::

This remarkable formula connects:
- A macroscopic observable (Hall conductance, measured in volts and currents)
- A microscopic quantum topological invariant (Chern number, counting how many times Berry curvature "winds" around momentum space)

The result is that the Hall conductance is quantized in units of $e^2/h$ and immune to disorder (up to localization effects).

:::{admonition} Derivation: Topological Hall Conductance
:class: dropdown

When a magnetic field is applied perpendicular to a 2D electron system, the kinetic energy operator becomes noncommuting in different directions. Electrons cannot move freely in the field direction (they form Landau levels), but within each band, Berry curvature encodes an effective "anomalous velocity" perpendicular to the applied electric field.

The transverse conductance arises entirely from this Berry-curvature-induced anomalous velocity. Integrating over all occupied states in all occupied bands, the current is proportional to the total Chern number.

:::

### Topological vs. Geometric Invariants

It is crucial to distinguish:

**Geometric (Berry) phases:**
- Path-dependent
- Change when the path changes
- Observable in open-path interferometry
- Example: $\gamma_n = \oint_C \mathcal{A}_n \cdot \mathrm{d}\boldsymbol{R}$ depends on the specific path $C$

**Topological invariants (Chern numbers):**
- Path-independent (for closed loops)
- Invariant under smooth deformations of the system
- Characterize global structure of the Hilbert space bundle
- Example: $C_n = \frac{1}{2\pi}\int_{\text{BZ}} \mathcal{F}_{xy}^{(n)}\,\mathrm{d}^2k$ depends only on the band structure, not on how we parametrize it

A Chern number is an *integrated* geometric property — it sums up the Berry curvature to yield an integer that encodes global topology.

### Bulk-Boundary Correspondence (Index Theorem)

One of the most profound results in topological band theory:

**Index Theorem**: If a 2D system with bulk Chern number $C_n \neq 0$ is terminated at a boundary, there must exist **chiral edge states** at that boundary.

The number of gapless one-way propagating edge modes equals $|C_n|$:
- If $C = +1$: one chiral edge state propagating clockwise
- If $C = -1$: one chiral edge state propagating counterclockwise
- If $C = 0$: no necessary edge states (generic case)

These edge states are **topologically protected** — they cannot be removed by disorder or perturbations without closing the bulk band gap.

:::{admonition} Physical Explanation
:class: note

Imagine "cutting" the 2D system along a line. The Berry curvature in the bulk represents a kind of "circulation" of electron momentum. At the boundary, this circulation must be interrupted. The result is a current that flows along the edge, carried by states that live at the boundary.

These edge currents are one-directional and robust because reversing them would require changing the global Chern number — an energetically expensive process involving gap closure.

:::

### Time-Reversal Symmetry and Topological Insulators

If a system has **time-reversal symmetry** (T-symmetry), then:

$$


\hat{T}\hat{H}\hat{T}^{-1} = \hat{H}

$$


For fermionic systems, T-symmetry requires that bands come in pairs with opposite Chern numbers:

$$


C_n(\boldsymbol{k}) = -C_n(-\boldsymbol{k})

$$


Therefore, the total Chern number for all occupied bands is forced to be **zero**, and $C = 0$ does not predict edge states.

However, TR-symmetric systems can be **topologically nontrivial** via a different integer invariant: the **$\mathbb{Z}_2$ topological invariant**. This predicts an even number of protected edge modes that are resilient to backscattering but not one-directional.

Such systems are called **topological insulators**.

### Winding Numbers in 1D Chains

In one-dimensional systems (e.g., the **SSH model** — Su-Schrieffer-Heeger model of polyacetylene), the analog of the Chern number is the **winding number**:

$$


w = \frac{1}{2\pi}\oint_0^{2\pi} \arg(\det[\hat{H}(\boldsymbol{k})])\,\mathrm{d}k \in \mathbb{Z}

$$


This integer also cannot change without closing the band gap. A non-zero winding number predicts **zero-energy edge states** (bound states at the ends of the chain).

For the SSH model:
- Two sublattices (A and B) in a 1D chain
- Alternating hopping amplitudes $t_1$ and $t_2$
- Winding number $w = \mathrm{sign}(t_2 - t_1)$ predicts which sublattice hosts the edge state

This is the 1D version of bulk-boundary correspondence.

### Key Physical Consequences of Topological Structure

1. **Quantized Responses**: Hall conductance, thermal Hall effect, etc., are quantized integers determined by Chern numbers.

2. **Robust Edge States**: Gapless boundary modes exist and are protected against disorder, defects, and weak perturbations.

3. **Immune to Disorder**: As long as the band gap remains open, disorder cannot change the Chern number or destroy the topological protection.

4. **Observable Signatures**:
   - Quantized transverse conductance
   - Chiral edge currents
   - Anomalous transport (anomalous Hall effect)
   - Edge magnetization and thermal transport

5. **Applications**: Topological insulators and materials promise fault-tolerant electronics, quantum computing platforms, and robust quantum devices.
