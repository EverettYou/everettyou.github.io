## Lecture Notes

The quantum Hall effect reveals that the Hall conductance is quantized in integer multiples of e²/h, independent of microscopic details or disorder—a universal transport coefficient set by fundamental constants alone. This topological quantization arises because the filled Landau level supports only a fixed number of current-carrying states, regardless of sample impurity or geometry. The integer quantum Hall effect is a triumph of topological thinking in condensed matter physics, and a window into how topology shapes quantum phenomena.
:::{admonition} Integer Quantum Hall Effect is Topological
:class: important

The quantized Hall resistance $\rho_{xy} = h/(ne^2)$ (where $n$ is an integer) is exact and independent of sample geometry, purity, or material details. This topological protection is why IQHE is so robust.
:::


### Filling Factor and Quantum Hall Plateaus

In a 2D electron gas in a strong perpendicular magnetic field, the key parameter is the **filling factor**:

$$

\nu = \frac{N_e}{N_\phi}

$$

where $N_e$ is the number of electrons and $N_\phi = \Phi/\Phi_0$ is the number of flux quanta through the system.

The filling factor tells us how many Landau levels are occupied. For $\nu = 1$, the lowest Landau level is completely filled. For $\nu = 2$, both the lowest and first excited Landau levels are filled, etc.

As we vary the magnetic field (keeping electron number fixed), the filling factor changes. When $\nu$ is an integer, the Hall conductance exhibits remarkable behavior.

### Integer Quantum Hall Conductance

The **quantum Hall effect** occurs when a 2D electron gas in a perpendicular magnetic field is subject to an applied electric field. An electric field in the $x$-direction drives a perpendicular (Hall) current in the $y$-direction. The Hall conductance is:

$$

\sigma_H = \frac{I_y}{E_x}

$$

**Von Klitzing's experimental discovery (1980)**: When an integer number $\nu$ of Landau levels are completely filled, the Hall conductance is **exactly quantized**:

$$

\sigma_H = \nu\frac{e^2}{h}

$$

This quantization is:
- **Exact**: Measured to parts per billion precision
- **Universal**: Independent of material properties, sample shape, even disorder strength
- **Fundamental**: Determined by only two fundamental constants—$e$ and $h$—with no dependence on sample parameters

For a single filled Landau level ($\nu = 1$):

$$

\sigma_H = \frac{e^2}{h} \approx 77.5\, \mu S

$$

This quantity $e^2/h$ is so fundamental it's called the **conductance quantum**. The integer quantum Hall effect is one of the most precisely measured phenomena in physics, rivaling even atomic spectroscopy in accuracy.

:::{admonition} Derivation: Transport Quantization from Edge States
:class: dropdown note


In a magnetic field $B$ applied perpendicular to a 2D electron system, Landau levels form at energies $E_n = \hbar\omega_c(n + 1/2)$ with degeneracy $N_\phi = \Phi_{\text{total}}/\Phi_0$ per level, where $\Phi_0 = h/e$ is the flux quantum.

When $\nu$ Landau levels are completely filled, there are $N_e = \nu N_\phi$ electrons.

The key insight: In the bulk, Landau level states are localized by disorder. At the sample **boundary**, the confining potential breaks the degeneracy and creates **extended edge states** that propagate freely along the boundary. When $\nu$ levels are filled, exactly $\nu$ edge state branches carry current, each with conductance $e^2/h$.

The total Hall conductance is the sum:

$$

\sigma_H = \nu \times \frac{e^2}{h}

$$

This transport is **topologically protected**: the edge states must exist as long as the energy gap separating filled from empty Landau levels persists, regardless of edge details or disorder. This protection is the reason for the remarkable precision.

:::

### Origin of Quantization: Edge States

The key to understanding quantum Hall quantization is the role of **edge states**.

**Bulk states**: In the interior of the sample, Landau level states are highly degenerate. Disorder (impurities, defects) localizes these states into immobile clusters. Bulk electrons cannot contribute to transport—they are trapped in localized orbitals.

**Edge states**: At the physical boundary of the sample, the confining potential bends the Landau level structure. States that would be localized in the bulk become **extended along the edge**—they propagate freely along the boundary like a 1D wire.

When $\nu$ Landau levels are filled:
- The $\nu$ lowest edge state branches each propagate along the boundary
- Each branch carries conductance $e^2/h$ independently
- The total conductance is the sum: $\sigma_H = \nu e^2/h$

This edge transport is **topologically protected**: the states must exist as long as the Landau level gap is maintained, regardless of edge geometry or disorder. This is why the quantum Hall effect is robust—disorder that localizes bulk states cannot destroy the edge transport.

### Charge Pumping

As the magnetic field increases adiabatically, the Landau level structure evolves. Charges are "pumped" from the bulk to the edge—a quantum transfer of charge across an insulating bulk. This reveals the topological nature: the quantum Hall state transports charge via winding numbers and topological defects, not via conventional current paths.

### Linear Response Theory

In linear response to a weak applied electric field, the current-field relation is governed by the **conductivity tensor**:

$$

J_i = \hat{\sigma}_{ij}E_j

$$

In a magnetic field perpendicular to a 2D system, the conductivity tensor has the form:

$$

\hat{\sigma}_{ij} = \hat{\sigma}_{xx}\delta_{ij} + \hat{\sigma}_{xy}\epsilon_{ij}

$$

where $\hat{\sigma}_{xx}$ is the longitudinal (dissipative) conductivity and $\hat{\sigma}_{xy} = \sigma_H$ is the Hall conductivity.

**In the quantum Hall regime**, when a Landau level is completely filled, something remarkable happens:

$$

\hat{\sigma}_{xx} \approx 0 \quad (\text{perfect insulator})

$$

$$

\hat{\sigma}_{xy} = \nu\frac{e^2}{h} \quad (\text{perfectly quantized})

$$

The system acts as a perfect insulator in the direction of the applied field but shows **perfect quantized conductance** perpendicular to both field and current. This is impossible in classical physics—it arises purely from the topological structure of Landau levels and the existence of extended edge states.

:::{admonition} Derivation: Conductivity Tensor and Transport Coefficients
:class: dropdown note


In the quantum Hall regime, the bulk is insulating (localized states), and all transport flows along the $\nu$ edge branches. Each edge branch behaves like a 1D quantum wire with quantized conductance $e^2/h$.

A particle with charge $e$ traveling along an edge picks up an energy $eE$ from the applied field. In steady state, the current carried by one edge branch is:

$$

I_{\text{edge}} = \frac{e^2 E}{h}

$$

so $\sigma_{\text{edge}} = e^2/h$.

With $\nu$ filled Landau levels, there are $\nu$ independent edge branches, so:

$$

\hat{\sigma}_{xy} = \nu \times \frac{e^2}{h}

$$

The longitudinal conductivity is suppressed because the bulk is insulating and disorder cannot create midgap states that would short across the sample.

:::

### Topological Nature: Chern Number

The quantization of Hall conductance is fundamentally **topological**—it reflects a topological invariant called the **Chern number** of the electron band.

The Chern number is defined in momentum space as:

$$

C = \frac{1}{2\pi}\int_{\text{BZ}}d^2k\, \Omega(\boldsymbol{k})

$$

where $\Omega(\boldsymbol{k})$ is the Berry curvature in momentum space and the integral is over the Brillouin zone.

For Landau levels:
- The Chern number of the $n$-th Landau level is $C_n = 1$ (or $-1$ depending on gauge convention)
- When $\nu$ Landau levels are filled, the total Chern number is $C_{\text{tot}} = \nu$
- The Hall conductance is directly related: $\sigma_H = C_{\text{tot}} \cdot e^2/h$

The Chern number is a **topological invariant**—it cannot change under continuous deformations of the band structure that don't close the gap. This explains why quantization is so robust.

### Robustness to Disorder

A remarkable feature of the quantum Hall effect is its insensitivity to disorder.

**Why disorder doesn't matter**:
1. Bulk Landau level states are already localized by the magnetic field itself (cyclotron radius ≲ magnetic length)
2. Weak disorder cannot delocalize already-localized bulk states
3. Edge states remain extended and protected by the topology of the bulk
4. The topological protection persists as long as the bulk gap (Landau level spacing) is maintained

This is why the IQHE was so surprising and important—it established that macroscopic quantum effects can be completely universal, independent of material quality. This principle underlies modern quantum metrology and topological quantum computation.

### Fractional Quantum Hall Effect

At fractional filling factors like $\nu = 1/3, 2/5, 3/7, \ldots$, additional quantized Hall plateaus appear! These are due to many-body effects (electron-electron interactions):

- Electrons within a partially filled Landau level interact via Coulomb repulsion
- Correlations can create new incompressible quantum states
- Each fractional state has its own topological quantum number

The fractional quantum Hall effect is beyond the single-particle Landau level picture but follows the same topological principles.

### Summary

- **Filling factor**: $\nu = N_e/N_\phi$ counts electrons per flux quantum; integer $\nu$ labels integer QHE plateaus.
- **Hall conductance**: $\sigma_H = \nu e^2/h$ is exactly quantized when $\nu$ Landau levels are filled.
- **Universal and exact**: Quantization depends only on fundamental constants, not material properties.
- **Edge states**: Carry all current; protected by topology; bulk states are localized and inactive.
- **Topological protection**: Robust against disorder, temperature, geometry—as long as Landau level gap persists.
- **Charge pumping**: Increasing flux by $\Phi_0$ pumps exactly $\nu e$ out of the system.
- **Chern number**: Topological invariant that directly equals $\nu$ for integer quantum Hall states.
- **Landau level spacing**: $\hbar\omega_c$ must be large compared to temperature and disorder strength.
- **Experimental precision**: IQHE measured to better than 1 part in $10^9$; used to define resistance standard.
