## Lecture Notes

The quantum Hall effect (QHE) is one of the most celebrated discoveries in condensed matter physics. It demonstrates that in a strong perpendicular magnetic field, the Hall conductance becomes *quantized* in units of $e^2/h$, achieving values independent of microscopic details like sample purity and geometry. This unexpected robustness points to deep topological structure underlying the effect.

### Integer Quantum Hall Effect (IQHE)

#### Experimental Setup and Observation

A 2D electron gas (a thin layer of free electrons confined to move in a plane) is placed in a perpendicular magnetic field $B$. Current $I$ flows in one direction (the $x$-direction), and the transverse voltage $V_y$ (perpendicular to current) is measured.

In a classical conductor, Ohm's law gives $V_y = R_H I$ where $R_H$ is the Hall resistance. The classical Hall coefficient is:

$$


R_H^{\text{cl}} = \frac{B}{ne}

$$


Crucially, $R_H^{\text{cl}}$ is continuous in $B$ and depends on the carrier density $n$.

**The quantum surprise**: In a strong magnetic field at low temperature, $R_H$ jumps to quantized values:

:::{admonition} Integer Quantum Hall Conductance
:class: important

$$


\sigma_{xy} = \nu \frac{e^2}{h}, \quad R_H = \frac{h}{\nu e^2}

$$


where $\nu = 1, 2, 3, \ldots$ is the **filling factor**, an integer.

The conductance is *exactly* quantized—independent of disorder, sample geometry, and other microscopic details.
:::

#### Filling Factor and Magnetic Length

The filling factor is the ratio of electrons to flux quanta:

:::{admonition} Filling Factor
:class: important

$$


\nu = \frac{N_e}{N_\phi}

$$


where $N_e$ is the number of electrons and $N_\phi = BA/\Phi_0 = BA\cdot e/h$ is the number of flux quanta piercing the sample.

For integer $\nu$, exactly $\nu$ Landau levels are filled.
:::

When $\nu$ is an integer, the Fermi level lies in a gap between filled and empty Landau levels. This gap (the **cyclotron gap**) has size $\hbar\omega_c$ and is crucial for the robustness of the effect.

#### Hall Conductance Derivation

Consider a filled Landau level with $N_{\phi}$ electrons. Each electron carries charge $-e$. A current $I$ flowing through the sample means charges move with drift velocity. The Hall field arises from the Lorentz force:

$$


\boldsymbol{F} = -e\boldsymbol{v} \times \boldsymbol{B}

$$


At steady state, the electric field $E_y$ balances this force to prevent further acceleration. The relation between Hall field and current gives:

:::{admonition} Derivation: Hall Conductance
:class: dropdown

For a filled Landau level in a Hall bar of width $W$ and length $L$:
- Current $I$ means $I/e$ electrons per unit time cross a transverse section.
- These electrons are deflected by $\boldsymbol{B}$, creating a charge accumulation on one edge.
- The resulting electric field $E_y$ creates a force that exactly cancels the Lorentz force for the drifting electrons.

The Hall voltage is $V_y = E_y \cdot W$. From the balance of forces and the number of filled levels:

$$


\sigma_{xy} = \frac{I}{V_y/L} = \frac{e^2 h}{h} \cdot \nu = \nu \frac{e^2}{h}

$$


The result depends only on $\nu$ (the number of filled Landau levels) and fundamental constants—not on $B$, $n$, or sample properties.
:::

### Robustness and Topological Origin

#### Why Disorder Doesn't Destroy Quantization

Intuitively, disorder should spoil any delicate quantum effect. Yet the QHE conductance is observed to be quantized to parts per billion precision, even in very dirty samples.

The key is the **Landau level gap**: Disorder cannot mix states within a single Landau level if the energy cost is $\hbar\omega_c$. Electrons in a filled level cannot scatter into empty states above the gap. Thus:

:::{admonition} Disorder Robustness
:class: important

- The quantized Hall conductance is **topologically protected**: it depends only on the topology of the filled band structure, not microscopic details.
- Disorder can localize states deep in the Landau level bulk, but **edge states** remain delocalized.
- Edge states carry the current; their number is fixed by the topology, independent of disorder.
:::

#### Edge States

A real sample has boundaries. Near the boundary, the potential energy is not uniform—it rises toward the edge (to confine electrons). This confining potential shifts Landau level energies, causing them to fan upward toward the boundary. The result: some states near the edge have energy within the bulk gap.

These **edge states** are topological: they are guaranteed to exist at the boundary of a filled Landau level, and they remain intact even with disorder.

:::{admonition} Edge State Properties
:class: note

- Edge states are chiral: they propagate in one direction only (set by $\boldsymbol{E} \times \boldsymbol{B}$ drift).
- Each edge state carries exactly one quantum $e^2/h$ of conductance.
- The total Hall conductance equals the number of chiral edge states times $e^2/h$.
- Disorder cannot backscatter an electron along the edge without energy cost $\sim \hbar\omega_c$.
:::


:::{admonition} Discussion
:class: dropdown warning

The statement "disorder cannot mix states within a single Landau level if the energy cost is $\hbar\omega_c$" is powerful, but requires careful thinking. Consider an electron in a filled Landau level (all states occupied) with a small random potential $V(\boldsymbol{r})$ due to impurities.

(a) If the electron scatters off an impurity, can it scatter into an empty state above the current Landau level (i.e., into the next Landau level)? What is the minimum energy it must gain to do so?

(b) Within the current filled Landau level, are there empty states available into which the electron could scatter? If all states are filled, what does the Pauli exclusion principle forbid?

(c) Explain carefully why "the quantized Hall conductance is topologically protected": How does the absence of scattering channels (due to the Landau level gap and Pauli blocking) guarantee that edge states remain delocalized even with strong disorder?

(d) What happens if the disorder potential has regions where $V(\boldsymbol{r}) > \hbar\omega_c$? Can such strong disorder create a scattering channel into the next Landau level?
:::

### Laughlin's Wavefunction and Fractionality

For the integer case ($\nu$ integer), all electrons fit neatly into $\nu$ filled Landau levels. But what happens at fractional fillings, like $\nu = 1/3$?

At fractional fillings, electron-electron interactions become essential. The ground state is a highly entangled many-body state with **long-range order**. Laughlin proposed an elegant wavefunction for $\nu = 1/m$ (where $m$ is odd):

$$


\Psi_{\text{Laughlin}} = \prod_{i<j} (z_i - z_j)^m \exp\left(-\frac{1}{4\ell_B^2}\sum_i |z_i|^2\right)

$$


This wavefunction describes electrons with filling $\nu = 1/m$. Its key features are:

:::{admonition} Laughlin Wavefunction Insights
:class: note

- The factor $(z_i - z_j)^m$ enforces that electrons repel: the closer two electrons, the more suppressed the wavefunction.
- The state is a **quantum fluid**—incompressible, with a gap to excitations.
- Excitations are **fractional quasiparticles** carrying charge $e/m$ (and fractional statistics).
- Laughlin's picture extends to the fractional quantum Hall effect (FQHE), where $\nu$ is a rational fraction.
:::

### Experimental Significance and Modern Applications

The quantum Hall effect had enormous impact:

:::{admonition} Historical and Practical Importance
:class: important

- **Discovery (1980)**: Von Klitzing discovered the IQHE and won the Nobel Prize in 1985.
- **Precision metrology**: QHE provides a standard for electrical resistance, independent of microscopic details. It defines the ohm in the SI system.
- **Topological quantum matter**: QHE is the prototypical topological phase—a state of matter with no classical analog, protected by global constraints rather than symmetry.
- **Modern research**: QHE connections to topological insulators, topological superconductors, and anyons continue to drive fundamental physics and quantum information science.
:::

### Summary

:::{admonition} Key Results: Quantum Hall Effect
:class: important

- **Integer QHE**: Hall conductance $\sigma_{xy} = \nu e^2/h$ is exactly quantized at integer filling $\nu$.
- **Filling factor**: $\nu = N_e/N_\phi$; integer $\nu$ means $\nu$ Landau levels are completely filled.
- **Robustness**: Quantization is independent of disorder, sample geometry, and microscopic details.
- **Origin**: Landau level gap and topological edge states ensure current is carried by delocalized chiral edge modes.
- **Fractional QHE**: At fractional fillings, interactions create incompressible quantum fluids with fractional charge excitations.
- **Significance**: Canonical example of topological phases; defines the ohm in SI metrology.
:::
