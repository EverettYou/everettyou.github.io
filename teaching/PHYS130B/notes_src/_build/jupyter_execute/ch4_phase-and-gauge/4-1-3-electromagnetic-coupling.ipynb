## Lecture Notes

Electromagnetism emerges naturally as the simplest possible gauge theory: a local $U(1)$ symmetry that requires the existence of massless gauge bosons (photons) coupling to charged particles with universal strength. In this section, we uncover the minimal coupling prescription that encodes electromagnetic interactions in the quantum Hamiltonian, and show how the classical Maxwell equations emerge from demanding gauge covariance. This is the prototype for understanding all fundamental forces.

### Charged Particle Hamiltonian with Electromagnetic Coupling

The Hamiltonian for a charged particle of charge $q$ and mass $m$ in electromagnetic fields is:

$$

\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi

$$

where $\hat{\boldsymbol{p}} = -i\hbar\nabla$ is the canonical momentum operator, $\boldsymbol{A}(x, t)$ is the vector potential (with magnetic field $\boldsymbol{B} = \nabla \times \boldsymbol{A}$), and $\Phi(x, t)$ is the scalar potential (with electric field $\boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}$). This is the **minimal coupling** form: we simply replace the canonical momentum $\hat{\boldsymbol{p}}$ with the covariant momentum $\hat{\boldsymbol{p}} - q\boldsymbol{A}$ in the kinetic energy term. Expanding the kinetic energy carefully in the Coulomb gauge $\nabla \cdot \boldsymbol{A} = 0$ gives:

$$

\hat{H} = \frac{\hat{\boldsymbol{p}}^2}{2m} - \frac{q}{m}\boldsymbol{A} \cdot \hat{\boldsymbol{p}} + \frac{q^2}{2m}\boldsymbol{A}^2 + q\Phi

$$

The three terms after $\hat{\boldsymbol{p}}^2/(2m)$ represent: the linear coupling to the vector potential (dominates for weak fields), the quadratic term (important for strong fields), and the scalar potential interaction respectively. Minimal coupling is the simplest way to introduce interactions with gauge fields—it ensures gauge invariance while introducing the minimum number of new terms.

:class: dropdown note

The Hamiltonian for a charged particle of charge $q$ and mass $m$ in electromagnetic fields $\boldsymbol{E}$ and $\boldsymbol{B}$ is:

$$

\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi

$$

where:
- $\hat{\boldsymbol{p}} = -i\hbar\nabla$ is the canonical momentum operator
- $\boldsymbol{A}(x, t)$ is the vector potential (magnetic field: $\boldsymbol{B} = \nabla \times \boldsymbol{A}$)
- $\Phi(x, t)$ is the scalar potential (electric field: $\boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}$)

This is the **minimal coupling** form, derived by the replacement $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$ in the free-particle Hamiltonian $\hat{H}_0 = \hat{\boldsymbol{p}}^2/(2m)$.

Expanding the kinetic energy term:

$$

\frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 = \frac{1}{2m}[\hat{\boldsymbol{p}}^2 - q(\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}}) + q^2\boldsymbol{A}^2]

$$

Since $[\boldsymbol{A}, \boldsymbol{p}]$ generally doesn't vanish (because $\boldsymbol{A}$ varies in space), we need to be careful with the order of operators. The symmetric form is:

$$

\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}} = 2\boldsymbol{A} \cdot \hat{\boldsymbol{p}} + (\nabla \cdot \boldsymbol{A})

$$

In the **Coulomb gauge** where $\nabla \cdot \boldsymbol{A} = 0$, this simplifies to:

$$

\hat{H} = \frac{\hat{\boldsymbol{p}}^2}{2m} - \frac{q}{m}\boldsymbol{A} \cdot \hat{\boldsymbol{p}} + \frac{q^2}{2m}\boldsymbol{A}^2 + q\Phi

$$

:::

### Canonical vs Kinetic Momentum

Two distinct momentum operators appear in the electromagnetic theory and play different roles:

**Canonical momentum**: $\hat{\boldsymbol{p}} = -i\hbar\nabla$ is the generator of spatial translations and satisfies the fundamental commutation relation $[\hat{p}_i, \hat{x}_j] = -i\hbar\delta_{ij}$. It couples to forces and is related to the Noether current for translation symmetry.

**Kinetic momentum** (mechanical momentum): $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ appears in the kinetic energy term and determines the actual particle velocity: $\hat{\boldsymbol{v}} = \hat{\boldsymbol{\pi}}/m$. This is the momentum that determines how fast the particle is actually moving.

**Key difference**: In a magnetic field, the canonical momentum $\hat{\boldsymbol{p}}$ is generally **not conserved** because the electromagnetic force acts on the particle. However, kinetic momentum is also not separately conserved in a time-dependent field. The important distinction is that kinetic momentum directly measures the physical particle velocity, while canonical momentum is more abstract but has better symmetry properties.

The kinetic momentum is never separately conserved in a time-dependent field.

### Gauge Covariance of the Hamiltonian

Under a gauge transformation where the wavefunction transforms as $\psi \to \psi' = e^{i\chi}\psi$ and the potentials transform as $\boldsymbol{A} \to \boldsymbol{A}' = \boldsymbol{A} - \frac{1}{q}\nabla\chi$ and $\Phi \to \Phi' = \Phi + \frac{1}{q}\partial_t\chi$, the Hamiltonian transforms covariantly. That is, the transformed Hamiltonian applied to the transformed wavefunction gives the gauge-transformed result:

$$

\hat{H}'\psi' = e^{i\chi}\hat{H}\psi

$$

This gauge covariance is **crucial**: it ensures that the eigenvalues (measurable energies) and eigenstates of the Hamiltonian are gauge-invariant. Physically, this means that no matter which gauge we choose to work in, the quantum dynamics and predictions are identical. This is a profound principle: the theory is independent of our choice of gauge.

:::{admonition} Verification: Gauge Covariance Calculation
:class: dropdown note

Under a gauge transformation:

$$

\psi \to \psi' = e^{i\chi}\psi

$$

$$

\boldsymbol{A} \to \boldsymbol{A}' = \boldsymbol{A} - \frac{1}{q}\nabla\chi

$$

$$

\Phi \to \Phi' = \Phi + \frac{1}{q}\partial_t\chi

$$

The Hamiltonian transforms as:

$$

\hat{H}'\psi' = (\frac{1}{2m}[(\hat{\boldsymbol{p}} - q\boldsymbol{A}') + (q\Phi')]e^{i\chi}\psi

$$

Substituting the gauge transformation rules and using the chain rule carefully shows:

$$

\hat{H}'\psi' = e^{i\chi}\hat{H}\psi

$$

The Hamiltonian is **gauge covariant**—the transformed Hamiltonian applied to the transformed wavefunction gives the gauge-transformed result of the original Hamiltonian. This ensures that observables (eigenvalues of operators) are gauge-invariant.

:::

### Quantum Lorentz Force

The equations of motion for expectation values in quantum mechanics follow from the Heisenberg equation. For the velocity, we compute:

$$

\frac{d\langle\boldsymbol{x}\rangle}{dt} = \langle\boldsymbol{v}\rangle = \frac{1}{m}\langle\boldsymbol{\pi}\rangle

$$

For the canonical momentum, the commutator with the Hamiltonian gives:

$$

\frac{d\langle\boldsymbol{p}\rangle}{dt} = q\langle\boldsymbol{E}\rangle + q\langle\boldsymbol{v} \times \boldsymbol{B}\rangle

$$

This is exactly the **Lorentz force law** in quantum form:

$$

m\frac{d\langle\boldsymbol{v}\rangle}{dt} = q(\langle\boldsymbol{E}\rangle + \langle\boldsymbol{v} \times \boldsymbol{B}\rangle)

$$

**Remarkable fact**: The quantum equations of motion reduce to the classical Lorentz force law when we take expectation values! This is a manifestation of the **correspondence principle**, showing that quantum mechanics correctly reproduces classical mechanics in the appropriate limit. The covariant formulation of quantum mechanics guarantees this agreement at the level of expectation values.

:::{admonition} Derivation: Heisenberg Equations and Lorentz Force
:class: dropdown note

Using the Heisenberg equation of motion for the position operator:

$$

\frac{d\hat{\boldsymbol{x}}}{dt} = \frac{1}{i\hbar}[\hat{\boldsymbol{x}}, \hat{H}]

$$

The commutator yields:

$$

\frac{d\langle\boldsymbol{x}\rangle}{dt} = \langle\boldsymbol{v}\rangle = \frac{1}{m}\langle\boldsymbol{\pi}\rangle

$$

For the canonical momentum:

$$

\frac{d\hat{\boldsymbol{p}}}{dt} = \frac{1}{i\hbar}[\hat{\boldsymbol{p}}, \hat{H}]

$$

After careful computation:

$$

\frac{d\langle\boldsymbol{p}\rangle}{dt} = q\langle\boldsymbol{E}\rangle + q\langle\boldsymbol{v} \times \boldsymbol{B}\rangle

$$

This is exactly the **Lorentz force law** in quantum form:

$$

m\frac{d\langle\boldsymbol{v}\rangle}{dt} = q\langle\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B}\rangle

$$

**Remarkable fact**: The quantum equations of motion reduce to the classical Lorentz force law when we take expectation values! This is a manifestation of the **correspondence principle**.

:::

### Gauge Invariance of Observables

A crucial principle of gauge theories: **all physical observables are gauge-invariant.** This statement summarizes why gauge freedom is merely redundancy, not ambiguity:

**Position and momentum expectations**: These are directly gauge-invariant—they're defined in position space and don't depend on the gauge choice:

$$

\langle\boldsymbol{x}\rangle' = \int (\psi')^*\boldsymbol{x}\psi'\, dx = \int(e^{-i\chi})\psi^*\boldsymbol{x}\psi(e^{i\chi})dx = \langle\boldsymbol{x}\rangle

$$

**Probability densities**: Also manifestly gauge-invariant:

$$

P'(x) = |\psi'|^2 = |e^{i\chi}|^2|\psi|^2 = |\psi|^2 = P(x)

$$

**More subtle observables** like angular momentum or current densities are also gauge-invariant if properly defined. The gauge potential $\boldsymbol{A}$ itself is **not observable** because different gauges give different values. Only the field strength $\boldsymbol{B} = \nabla \times \boldsymbol{A}$ (and correspondingly $\boldsymbol{E}$) are observable.

:::{admonition} Gauge Freedom as a Computational Tool
:class: important

The freedom to choose different gauges is not just a mathematical redundancy—it's a practical advantage. In solving the Schrödinger equation, we can choose the gauge that makes the math simplest for a particular problem. The Coulomb gauge, the Lorenz gauge, and radiation gauge each have advantages in different contexts. Different gauges may be convenient for different purposes, but all lead to identical physical predictions.

:::

### Path-Dependent Phases and the Aharonov-Bohm Effect

Although the vector potential $\boldsymbol{A}$ is not directly observable, **line integrals** of $\boldsymbol{A}$ are observable through quantum interference effects. A particle traveling from point $a$ to point $b$ along path $C$ acquires a quantum phase:

$$

\phi_C = \frac{q}{\hbar c}\int_C \boldsymbol{A} \cdot d\boldsymbol{l}

$$

Now consider two different paths $C_1$ and $C_2$ connecting the same two points. They enclose a region with magnetic flux $\Phi_B = \int_S (\nabla \times \boldsymbol{A}) \cdot d\boldsymbol{S}$. By Stokes' theorem:

$$

\phi_{C_1} - \phi_{C_2} = \frac{q}{\hbar c}\oint\boldsymbol{A} \cdot d\boldsymbol{l} = \frac{q\Phi_B}{\hbar c}

$$

**The crucial point**: The difference in phases between two paths is **completely gauge-invariant**—it depends only on the enclosed flux $\Phi_B$, not on the specific choice of vector potential $\boldsymbol{A}$ (which changes under gauge transformations). This is the origin of the Aharonov-Bohm effect: even in a region where $\boldsymbol{B} = 0$ (say, inside a solenoid shielded from the outside), the vector potential is nonzero, and particles can acquire observable phase shifts from path-dependent integrals of $\boldsymbol{A}$. This demonstrates that in quantum mechanics, the potentials $\boldsymbol{A}$ and $\Phi$ are **more fundamental** than the fields $\boldsymbol{E}$ and $\boldsymbol{B}$.

### Summary

- **Minimal coupling**: Charged particles couple to EM fields through $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$ in the Hamiltonian.
- **Hamiltonian form**: $\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi$ is gauge covariant.
- **Canonical momentum**: $\hat{\boldsymbol{p}} = -i\hbar\nabla$ is the generator of translations; it's frame-dependent.
- **Kinetic momentum**: $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ determines particle velocity; more physical but gauge-dependent.
- **Quantum Lorentz force**: Heisenberg equations give $m d\langle\boldsymbol{v}\rangle/dt = q\langle\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B}\rangle$.
- **Observable gauge invariance**: Physical quantities (probabilities, forces, energies) are independent of gauge choice.
- **Potentials are fundamental**: In quantum mechanics, $\boldsymbol{A}$ and $\Phi$ directly affect wavefunction evolution; classical electromagnetism privileges fields $\boldsymbol{E}$ and $\boldsymbol{B}$.
- **Path-dependent phases**: Line integrals of $\boldsymbol{A}$ encode observable interference—the foundation of Aharonov-Bohm effect.
