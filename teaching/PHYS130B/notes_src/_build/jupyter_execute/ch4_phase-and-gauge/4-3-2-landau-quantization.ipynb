## Lecture Notes

### Overview

Landau quantization describes the remarkable fate of a charged particle in a uniform magnetic field: continuous kinetic energy becomes quantized into discrete *Landau levels*, each with enormous degeneracy. The classical cyclotron orbit becomes a quantum harmonic oscillator. This is the foundation for understanding the quantum Hall effect and topological quantum phenomena.

### Semiclassical Quantization of Cyclotron Orbits

Before solving the full quantum problem, we can gain insight from Bohr-Sommerfeld quantization (§3.3). A classical charged particle in a magnetic field $\boldsymbol{B} = B\hat{z}$ executes circular orbits at the cyclotron frequency:

$$
\omega_c = \frac{eB}{m}
$$

The action integral around a closed orbit is:

$$
\oint \boldsymbol{p} \cdot \mathrm{d}\boldsymbol{r} = 2\pi m \omega_c r^2 = 2\pi \hbar(n + \tfrac{1}{2})
$$

where $r$ is the orbit radius. This gives quantized energies:

$$
E_n = \frac{1}{2}m\omega_c^2 r^2 = (n + \tfrac{1}{2})\hbar\omega_c
$$

:::{admonition} Bohr-Sommerfeld Quantization of Cyclotron Motion
:class: important

$$
E_n = \left(n + \tfrac{1}{2}\right)\hbar\omega_c \quad (n = 0, 1, 2, \ldots)
$$

This semiclassical result matches the exact quantum answer. The zero-point energy $E_0 = \frac{1}{2}\hbar\omega_c$ is purely quantum mechanical.
:::
:::{admonition} Derivation: Bohr-Sommerfeld Action Integral for Cyclotron Orbits
:class: dropdown information

For a charged particle in a uniform field $\boldsymbol{B} = B\hat{z}$, the classical orbit at energy $E$ is a circle of radius $r = v_\perp/\omega_c$ in the $xy$-plane, with kinetic momentum $p = mv_\perp$ tangent to the orbit.

The canonical momentum is $\boldsymbol{p} = m\boldsymbol{v} + e\boldsymbol{A}$. In the symmetric gauge $\boldsymbol{A} = \frac{B}{2}(-y, x, 0)$, the action integral around the orbit is:

$$

\oint \boldsymbol{p} \cdot \mathrm{d}\boldsymbol{r} = \oint m\boldsymbol{v} \cdot \mathrm{d}\boldsymbol{r} + e\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{r}

$$

**First term** (kinetic): For circular motion with $\vert\boldsymbol{v}\vert = v_\perp = \omega_c r$ tangent to the path and $\vert\mathrm{d}\boldsymbol{r}\vert = r\,\mathrm{d}\theta$:

$$

\oint m\boldsymbol{v} \cdot \mathrm{d}\boldsymbol{r} = m\omega_c r \cdot 2\pi r = 2\pi m\omega_c r^2

$$

**Second term** (gauge): By Stokes' theorem, $e\oint \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{r} = e\int (\nabla\times\boldsymbol{A})\cdot\mathrm{d}\boldsymbol{S} = eB\pi r^2$. However, the cyclotron orbit circulates in the direction *opposite* to the line integral orientation (for positive $q$ and $B$), so the enclosed flux picks up a sign:

$$

e\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{r} = -eB\pi r^2 = -\pi m\omega_c r^2

$$

**Total**: $\oint \boldsymbol{p}\cdot\mathrm{d}\boldsymbol{r} = 2\pi m\omega_c r^2 - \pi m\omega_c r^2 = \pi m\omega_c r^2$.

Applying the Bohr-Sommerfeld condition with Maslov index $\mu = 2$ (two turning points, as in any 1D oscillator):

$$

\pi m\omega_c r^2 = 2\pi\hbar\left(n + \tfrac{1}{2}\right)

$$

Since $E = \frac{1}{2}m\omega_c^2 r^2 = \frac{1}{2}m\omega_c \cdot (\omega_c r^2)$:

$$

E_n = \frac{1}{2}m\omega_c \cdot \frac{2\hbar(n + \tfrac{1}{2})}{m} = \left(n + \tfrac{1}{2}\right)\hbar\omega_c

$$
:::


### Quantum Hamiltonian in the Landau Gauge

The quantum Hamiltonian for a charged particle in a uniform magnetic field is:

$$
\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - e\boldsymbol{A})^2
$$

(In SI units; in Gaussian units use $q/c$ instead of $e$.)

For $\boldsymbol{B} = B\hat{z}$, we choose the **Landau gauge**:

$$
\boldsymbol{A} = (0, Bx, 0)
$$

This gauge breaks spatial translational symmetry in $x$ but reveals a crucial simplification: $\hat{p}_y$ commutes with the Hamiltonian and is a conserved quantum number.

The kinetic momentum components become:

$$
\hat{\pi}_x = \hat{p}_x, \quad \hat{\pi}_y = \hat{p}_y - eBx
$$

The Hamiltonian is:

$$
\hat{H} = \frac{\hat{p}_x^2}{2m} + \frac{(\hat{p}_y - eBx)^2}{2m}
$$

Using $[\hat{p}_x, x] = -\mathrm{i}\hbar$, we can shift the origin: let $X = x + \hat{p}_y/(eB)$. Then:

$$
\hat{H} = \frac{\hat{p}_x^2}{2m} + \frac{1}{2}m\omega_c^2 X^2
$$

This is a **harmonic oscillator in the $x$-direction** with frequency $\omega_c$. The variable $y$ (or equivalently $\hat{p}_y$) is a spectator: motion in $y$ is free.

:::{admonition} Landau Gauge Reduction
:class: note

In the Landau gauge, the Hamiltonian becomes a shifted harmonic oscillator:

$$
\hat{H} = \frac{\hat{p}_x^2}{2m} + \frac{1}{2}m\omega_c^2\left(x + \frac{\hat{p}_y}{eB}\right)^2
$$

The conserved momentum $\hat{p}_y$ labels degenerate states within each Landau level. The quantum number $n$ from the harmonic oscillator labels the Landau level.
:::


:::{admonition} Derivation: Kinetic-Momentum Commutator
:class: dropdown information

The kinetic momentum components are $\hat{\pi}_x = \hat{p}_x$ and $\hat{\pi}_y = \hat{p}_y - eBx$ in the Landau gauge. Their commutator is:

$$

[\hat{\pi}_x, \hat{\pi}_y] = [\hat{p}_x, \hat{p}_y - eBx] = [\hat{p}_x, \hat{p}_y] - eB[\hat{p}_x, x]

$$

Since canonical momenta commute ($[\hat{p}_x, \hat{p}_y] = 0$) and $[\hat{p}_x, x] = -\mathrm{i}\hbar$:

$$

[\hat{\pi}_x, \hat{\pi}_y] = 0 - eB(-\mathrm{i}\hbar) = \mathrm{i}\hbar eB

$$

This result is **gauge-invariant**. In a general gauge with $\boldsymbol{A} = (A_x, A_y)$:

$$

[\hat{\pi}_x, \hat{\pi}_y] = [\hat{p}_x - eA_x, \hat{p}_y - eA_y] = -e[\hat{p}_x, A_y] + e[A_x, \hat{p}_y - eA_y]

$$

$$

= -e(-\mathrm{i}\hbar\partial_x A_y) + e(\mathrm{i}\hbar\partial_y A_x) = \mathrm{i}\hbar e(\partial_x A_y - \partial_y A_x) = \mathrm{i}\hbar eB

$$

The noncommutativity of kinetic momentum components is a direct consequence of the magnetic field. It is the fundamental relation that makes the Landau problem equivalent to a harmonic oscillator, since $[\hat{\pi}_x, \hat{\pi}_y] = \mathrm{i}\hbar eB$ has the same structure as $[\hat{x}, \hat{p}] = \mathrm{i}\hbar$ with the replacement $\hbar \to \hbar eB$.
:::

### Landau Level Energies

The energy eigenvalues are determined by the harmonic oscillator quantum number $n$:

:::{admonition} Landau Level Energies and Degeneracy
:class: important

$$
E_n = \left(n + \tfrac{1}{2}\right)\hbar\omega_c \quad (n = 0, 1, 2, \ldots)
$$

All states with the same $n$ but different $\hat{p}_y$ belong to the $n$-th **Landau level**. There is no energy splitting within a level (to leading order).
:::

### Guiding Center Coordinates

A key insight is that quantum mechanics forbids simultaneously specifying both the $x$ and $y$ positions of a particle in a strong magnetic field. We can define the **guiding center coordinates**:

$$
\hat{X} = x + \frac{\hat{p}_y}{eB}, \quad \hat{Y} = y - \frac{\hat{p}_x}{eB}
$$

These represent the classical center of the cyclotron orbit. Their commutator is:

$$
[\hat{X}, \hat{Y}] = -\mathrm{i}\frac{\hbar}{eB} = -\mathrm{i}\ell_B^2
$$

:::{admonition} Guiding Center Noncommutativity
:class: important

$$
[\hat{X}, \hat{Y}] = -\mathrm{i}\ell_B^2 \quad \text{where} \quad \ell_B = \sqrt{\frac{\hbar}{eB}}
$$

This noncommutativity means that **guiding center positions cannot be simultaneously localized** to better than the magnetic length $\ell_B$. A quantum state occupies a minimum area of $\sim \ell_B^2$ in $(X, Y)$ space.
:::

:::{admonition} Derivation: Guiding Center Commutator
:class: dropdown information

Define the guiding center using kinetic momentum: $\hat{X} = x + \hat{\pi}_y/(eB)$ and $\hat{Y} = y - \hat{\pi}_x/(eB)$. We need:

$$

[\hat{X}, \hat{Y}] = \left[x + \frac{\hat{\pi}_y}{eB},\; y - \frac{\hat{\pi}_x}{eB}\right]

$$

Expanding:

$$

= [x, y] - \frac{1}{eB}[x, \hat{\pi}_x] + \frac{1}{eB}[\hat{\pi}_y, y] - \frac{1}{(eB)^2}[\hat{\pi}_y, \hat{\pi}_x]

$$

We evaluate each term. The coordinate commutator vanishes: $[x, y] = 0$. For the position-kinetic momentum commutators, $[x_i, \hat{\pi}_j] = \mathrm{i}\hbar\delta_{ij}$ (same as canonical, since $\hat{\pi}_j = \hat{p}_j - eA_j(\hat{\boldsymbol{x}})$ and $[x_i, A_j(\hat{\boldsymbol{x}})] = 0$). So $[x, \hat{\pi}_x] = \mathrm{i}\hbar$ and $[\hat{\pi}_y, y] = -[y, \hat{\pi}_y] = -\mathrm{i}\hbar$. The kinetic momentum commutator is $[\hat{\pi}_y, \hat{\pi}_x] = -[\hat{\pi}_x, \hat{\pi}_y] = -\mathrm{i}\hbar eB$.

Substituting:

$$

[\hat{X}, \hat{Y}] = 0 - \frac{\mathrm{i}\hbar}{eB} + \frac{(-\mathrm{i}\hbar)}{eB} - \frac{(-\mathrm{i}\hbar eB)}{(eB)^2}

$$

$$

= -\frac{\mathrm{i}\hbar}{eB} - \frac{\mathrm{i}\hbar}{eB} + \frac{\mathrm{i}\hbar}{eB} = -\frac{\mathrm{i}\hbar}{eB} = -\mathrm{i}\ell_B^2

$$

This result is gauge-invariant: it depends only on $B$, not on the choice of $\boldsymbol{A}$. The guiding center coordinates form a conjugate pair with effective Planck constant $\hbar_{\text{eff}} = \ell_B^2 = \hbar/(eB)$, analogous to $[x, p] = \mathrm{i}\hbar$.
:::


### Ladder Operators and Two Towers of States

The Landau level structure involves *two independent sets* of ladder operators:

1. **Landau level ladder operators** $\hat{a}, \hat{a}^\dagger$: These raise and lower the Landau level index $n$.

2. **Guiding center ladder operators** $\hat{b}, \hat{b}^\dagger$: These label distinct states *within* a Landau level (different guiding centers).

:::{admonition} Two Sets of Ladder Operators
:class: note

Define:

$$
\hat{a} = \sqrt{\frac{m\omega_c}{2\hbar}}(x + \tfrac{\hat{p}_y}{eB}) + \mathrm{i}\sqrt{\frac{1}{2\hbar m\omega_c}}\hat{p}_x
$$

$$
\hat{b} = \sqrt{\frac{m\omega_c}{2\hbar}}(y - \tfrac{\hat{p}_x}{eB}) + \mathrm{i}\sqrt{\frac{1}{2\hbar m\omega_c}}\hat{p}_y
$$

Both satisfy $[\hat{a}, \hat{a}^\dagger] = 1$ and $[\hat{b}, \hat{b}^\dagger] = 1$, but they commute with each other:

$$
[\hat{a}, \hat{b}] = [\hat{a}, \hat{b}^\dagger] = 0
$$

The Hamiltonian is:

$$
\hat{H} = \hbar\omega_c\left(\hat{a}^\dagger\hat{a} + \tfrac{1}{2}\right)
$$

States are labeled $\vert n, m \rangle$ where $\hat{a}^\dagger\hat{a}\vert n, m \rangle = n\vert n, m \rangle$ (Landau level) and $\hat{b}^\dagger\hat{b}\vert n, m \rangle = m\vert n, m \rangle$ (degeneracy index within the level).
:::

:::{admonition} Derivation: Ladder Operator Algebra from Kinetic Momentum and Guiding Center
:class: dropdown information

Define the cyclotron ladder operators as $\hat{a} = (\hat{\pi}_x + \mathrm{i}\hat{\pi}_y)/\sqrt{2\hbar eB}$ and $\hat{a}^\dagger = (\hat{\pi}_x - \mathrm{i}\hat{\pi}_y)/\sqrt{2\hbar eB}$.

**Canonical commutation:** Compute $[\hat{a}, \hat{a}^\dagger]$:

$$

[\hat{a}, \hat{a}^\dagger] = \frac{1}{2\hbar eB}[\hat{\pi}_x + \mathrm{i}\hat{\pi}_y,\; \hat{\pi}_x - \mathrm{i}\hat{\pi}_y]

$$

$$

= \frac{1}{2\hbar eB}\left([\hat{\pi}_x, \hat{\pi}_x] - \mathrm{i}[\hat{\pi}_x, \hat{\pi}_y] + \mathrm{i}[\hat{\pi}_y, \hat{\pi}_x] + [\hat{\pi}_y, \hat{\pi}_y]\right)

$$

$$

= \frac{1}{2\hbar eB}\left(0 - 2\mathrm{i}[\hat{\pi}_x, \hat{\pi}_y] + 0\right) = \frac{-2\mathrm{i} \cdot \mathrm{i}\hbar eB}{2\hbar eB} = 1

$$

**Hamiltonian in terms of** $\hat{a}$: Compute $\hat{a}^\dagger\hat{a} + \frac{1}{2}$:

$$

\hat{a}^\dagger\hat{a} = \frac{1}{2\hbar eB}(\hat{\pi}_x - \mathrm{i}\hat{\pi}_y)(\hat{\pi}_x + \mathrm{i}\hat{\pi}_y) = \frac{1}{2\hbar eB}(\hat{\pi}_x^2 + \hat{\pi}_y^2 + \mathrm{i}[\hat{\pi}_x, \hat{\pi}_y])

$$

$$

= \frac{\hat{\boldsymbol{\pi}}^2}{2\hbar eB} + \frac{\mathrm{i} \cdot \mathrm{i}\hbar eB}{2\hbar eB} = \frac{\hat{\boldsymbol{\pi}}^2}{2\hbar eB} - \frac{1}{2}

$$

Therefore $\hat{H} = \hat{\boldsymbol{\pi}}^2/(2m) = \hbar\omega_c(\hat{a}^\dagger\hat{a} + \frac{1}{2})$, using $\omega_c = eB/m$.

**Guiding center operators:** Similarly, $\hat{b} = \sqrt{eB/(2\hbar)}(\hat{\xi}_x - \mathrm{i}\hat{\xi}_y)$ with $[\hat{b}, \hat{b}^\dagger] = 1$ follows from $[\hat{\xi}_x, \hat{\xi}_y] = -\mathrm{i}\hbar/(eB)$ by the same algebra with $\hbar eB \to \hbar/(eB)$.

**Cross-commutators vanish:** Since $[\hat{\xi}_i, \hat{\pi}_j] = 0$ for all $i, j$ (the guiding center commutes with kinetic momentum — this can be verified by direct computation, e.g. $[\hat{\xi}_x, \hat{\pi}_x] = [x + \hat{\pi}_y/(eB), \hat{\pi}_x] = \mathrm{i}\hbar + (-\mathrm{i}\hbar eB)/(eB) = 0$), all linear combinations also commute: $[\hat{a}, \hat{b}] = [\hat{a}, \hat{b}^\dagger] = 0$.
:::


:::{admonition} Interpretation
:class: note

The appearance of two commuting sets of oscillators is profound:
- Quantum mechanics generates an internal structure (the $\hat{b}$ oscillator) from the noncommutativity of guiding center positions.
- States cannot be labeled by classical position $(X, Y)$; instead we need quantum numbers $n$ and $m$.
- This structure underlies the topological properties of the quantum Hall effect.
:::

### Complex Coordinates and Lowest Landau Level Wavefunctions

In the lowest Landau level ($n = 0$), wavefunctions have a special structure in complex coordinates $z = x + \mathrm{i}y$.

The lowest Landau level condition $\hat{a}\vert\psi\rangle = 0$ implies that the wavefunction must be **holomorphic** (analytic in $z$). The general form is:

$$
\psi_m(z, \bar{z}) = P_m(z) \exp\left(-\frac{\vert z\vert^2}{4\ell_B^2}\right)
$$

where $P_m(z)$ is a holomorphic polynomial of degree $m$ in $z$:

$$
P_m(z) = c_m z^m
$$

The Gaussian factor ensures normalizability; the polynomial factor labels the $m$ different degenerate states.

:::{admonition} Lowest Landau Level Wavefunctions
:class: important

$$
\psi_m(z, \bar{z}) = \frac{c_m}{\sqrt{\pi^{1/2}2^m m!\ell_B^{2m}}} z^m \exp\left(-\frac{\vert z\vert^2}{4\ell_B^2}\right)
$$

Each monomial $z^m$ corresponds to a distinct state within the $n=0$ level. The Gaussian width is set by $\ell_B$.
:::

:::{admonition} Why Holomorphic?
:class: dropdown information

The lowest Landau level condition is $\hat{a}\vert\psi\rangle = 0$, which expands to:

$$
\left[\sqrt{\frac{m\omega_c}{2\hbar}}(x + \mathrm{i}y) + \sqrt{\frac{1}{2\hbar m\omega_c}}\frac{\partial}{\partial z}\right]\psi = 0
$$

This is equivalent to $\frac{\partial\psi}{\partial\bar{z}} = 0$ (after rescaling), which means $\psi$ depends only on $z$, not $\bar{z}$. Thus $\psi$ is holomorphic.
:::

### Degeneracy of Landau Levels

The key fact is that many different guiding center positions (labeled by $m = 0, 1, 2, \ldots$) fit within the same Landau level. The number of distinct states in a Landau level is determined by the total magnetic flux through the sample.

For a sample of area $A$ in a uniform field $B$:

$$
\Phi_{\text{total}} = BA
$$

Each state occupies an area $\sim \ell_B^2$ in guiding center space. The degeneracy is:

$$
N_\phi = \frac{\Phi_{\text{total}}}{\Phi_0} = \frac{BA}{\Phi_0}
$$

where $\Phi_0 = h/e = 4.14 \times 10^{-15}$ T·m$^2$ is the **flux quantum**.

Equivalently:

$$
N_\phi = \frac{A}{\pi\ell_B^2}
$$

:::{admonition} Degeneracy Formula
:class: important

$$
N_\phi = \frac{BA}{\Phi_0} = \frac{A}{\pi\ell_B^2}
$$

For a macroscopic sample ($A \sim 1 \text{ mm}^2$) at $B \sim 1$ T, the degeneracy is $N_\phi \sim 10^{10}$—a macroscopic number. All these states have (nearly) identical energy.
:::

:::{admonition} Derivation: Landau Level Degeneracy from Flux Counting
:class: dropdown information

Each quantum state in a Landau level corresponds to a guiding center position. Since $[\hat{X}, \hat{Y}] = -\mathrm{i}\ell_B^2$, the minimum area per state in guiding center space is (by the uncertainty principle):

$$

\Delta X \cdot \Delta Y \geq \frac{\ell_B^2}{2}

$$

More precisely, the number of states is determined by the phase space volume. The guiding center forms a 2D phase space with effective $\hbar_{\text{eff}} = \ell_B^2$, so each state occupies area $2\pi\ell_B^2$ (the analog of $2\pi\hbar$ in 1D phase space). For a sample of area $A$:

$$

N_\phi = \frac{A}{2\pi\ell_B^2}

$$

Substituting $\ell_B^2 = \hbar/(eB)$:

$$

N_\phi = \frac{AeB}{2\pi\hbar} = \frac{AeB}{h} \cdot (2\pi) \cdot \frac{1}{2\pi} = \frac{BA}{h/e} = \frac{BA}{\Phi_0}

$$

where $\Phi_0 = h/e$ is the flux quantum. This is the total magnetic flux through the sample divided by one flux quantum: **each Landau level state "absorbs" exactly one flux quantum**.

Alternatively, in the Landau gauge, states are labeled by $k_y$ (the conserved $y$-momentum). The guiding center position is $X_0 = -\hbar k_y/(eB) = -k_y \ell_B^2$. For a sample of dimensions $L_x \times L_y$ with periodic boundary conditions in $y$: $k_y = 2\pi n_y/L_y$ with $n_y \in \mathbb{Z}$. The guiding center must lie within the sample: $0 \leq X_0 \leq L_x$, giving:

$$

0 \leq k_y \leq \frac{eBL_x}{\hbar} \quad \Rightarrow \quad N_\phi = \frac{L_x}{2\pi\ell_B^2/L_y} = \frac{L_xL_y}{2\pi\ell_B^2} = \frac{BA}{\Phi_0}

$$

confirming the flux-counting result from both gauges.
:::


:::{admonition} Discussion
:class: dropdown tip

In the quantum Hall regime, the degeneracy within a Landau level vastly exceeds the number of electrons:

(a) Suppose you have a 2D electron system with area $A = 1 \text{ mm}^2$ at $B = 5$ T, containing $N_e = 10^{11}$ electrons. What is the filling factor $\nu = N_e/N_\phi$? How many Landau levels are partially occupied?

(b) At very strong fields (say $B = 30$ T), why does the system become "incompressible"—unable to absorb additional electrons at a fixed Fermi energy?

(c) If disorder lifts the degeneracy by $\Delta E \sim \hbar\omega_c/1000$, does this destroy the quantum Hall effect?
:::

### Magnetic Length: The Fundamental Scale

The magnetic length emerges naturally from the commutation relations:

$$
\ell_B = \sqrt{\frac{\hbar}{eB}}
$$

It is the only length scale in the problem and determines the wavefunction size, the separation between lattice sites in guiding center space, and the energy scale of orbital physics.

:::{admonition} Magnetic Length (Recap)
:class: note

$$
\ell_B = \sqrt{\frac{\hbar}{eB}} \approx 25.66 \text{ nm} \times (1 \text{ T}/B)^{1/2}
$$

For $B = 1$ T: $\ell_B \approx 25.66$ nm  
For $B = 10$ T: $\ell_B \approx 8.1$ nm
:::

### Summary

- **Bohr-Sommerfeld quantization** of cyclotron orbits yields $E_n = (n + 1/2)\hbar\omega_c$, which matches the exact quantum result.
- **Landau gauge** $\boldsymbol{A} = (0, Bx, 0)$ reduces the problem to a shifted harmonic oscillator with $\hat{p}_y$ conserved.
- **Landau level energies** are $E_n = (n + 1/2)\hbar\omega_c$, independent of the guiding center position (to leading order).
- **Guiding center coordinates** satisfy $[\hat{X}, \hat{Y}] = -\mathrm{i}\ell_B^2$: positions are fundamentally fuzzy at the quantum scale.
- **Two ladder operators**: $\hat{a}, \hat{a}^\dagger$ label Landau levels; $\hat{b}, \hat{b}^\dagger$ label degenerate states within a level.
- **Lowest Landau level wavefunctions** are holomorphic polynomials $z^m$ times a Gaussian, reflecting the quantum constraint.
- **Degeneracy** $N_\phi = BA/\Phi_0 \sim 10^{10}$ for macroscopic samples: enormous and crucial for quantum Hall physics.
- **Magnetic length** $\ell_B = \sqrt{\hbar/(eB)}$ is the universal length scale characterizing quantum orbital motion.
