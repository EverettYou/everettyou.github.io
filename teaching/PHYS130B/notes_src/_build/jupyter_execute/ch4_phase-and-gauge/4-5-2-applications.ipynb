## Lecture Notes

### Born-Oppenheimer Approximation and Conical Intersections

In molecules, the nuclear coordinates $\boldsymbol{R}$ change slowly compared to electronic time scales. The **Born-Oppenheimer (BO) approximation** treats electronic degrees of freedom as adiabatically following the nuclear motion.

At each nuclear configuration $\boldsymbol{R}$, the electrons occupy an instantaneous eigenstate $|n(\boldsymbol{R})\rangle$ of the electronic Hamiltonian. As nuclei move, the electronic state acquires a Berry phase. This phase enters the nuclear equation of motion as a **geometric vector potential**:

$$

\tilde{\mathcal{A}}_{nm}(\boldsymbol{R}) = \langle m(\boldsymbol{R})|\nabla_{\boldsymbol{R}}|n(\boldsymbol{R})\rangle

$$

The nuclear wavefunction on the $n$-th BO surface couples to the $m$-th surface through this off-diagonal coupling. When $E_n(\boldsymbol{R}) = E_m(\boldsymbol{R})$ (a **conical intersection**), the BO approximation breaks down, and diabatic transitions become possible.

:::{admonition} Born-Oppenheimer Geometric Potential
:class: important

For a molecule undergoing adiabatic nuclear motion, the electronic Berry phase modifies the nuclear kinetic energy:

$$

\Delta V_{\text{geom}} = -\frac{\hbar^2}{2M}\left[(\nabla_{\boldsymbol{R}} \cdot \mathcal{A})^2 + 2\nabla_{\boldsymbol{R}} \cdot \mathcal{A}\right]

$$

where $\mathcal{A} = \mathrm{i}\langle n|\nabla_{\boldsymbol{R}}|n\rangle$ is the diagonal Berry connection and $M$ is the nuclear mass.

Near **conical intersections** where two electronic surfaces cross ($E_n = E_m$), the off-diagonal Berry connection $\langle m|\nabla_{\boldsymbol{R}}|n\rangle$ becomes singular, causing:
- Non-adiabatic transitions between electronic surfaces
- $\pm\pi$ phase jumps in the electronic wavefunction
- Branching of reaction pathways in photochemistry

:::

### The Aharonov-Bohm Effect as Berry Phase

In the original Aharonov-Bohm setup, an electron encircles a solenoid with magnetic flux $\Phi$. The electron's wavefunction acquires phase:

$$

\Delta\phi = \frac{e\Phi}{\hbar c}

$$

This can be reinterpreted as a Berry phase in *position space*. The "parameter" is the electron's position $\boldsymbol{r}$, and the flux tube acts as a "background" that generates an effective Berry connection:

$$

\mathcal{A}_{\text{eff}}(\boldsymbol{r}) = \frac\mathrm{e}{c\hbar}\boldsymbol{A}(\boldsymbol{r})

$$

where $\boldsymbol{A}$ is the vector potential. The phase accumulated around a closed loop is:

$$

\oint \mathcal{A}_{\text{eff}} \cdot \mathrm{d}\boldsymbol{r} = \frac\mathrm{e}{\hbar c}\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{r} = \frac{e\Phi}{\hbar c}

$$

This unifies the quantum Hall effect, band geometry, and electromagnetic interactions under a single geometric framework.

:::{admonition} Aharonov-Bohm Phase as Geometric Effect
:class: important

The Aharonov-Bohm effect demonstrates that magnetic flux can produce a measurable phase shift even when the electron never enters the region of non-zero magnetic field.

**Phase acquired by closed path around solenoid**:

$$

\phi_{\text{AB}} = \frac\mathrm{e}{c\hbar}\Phi_{\text{enclosed}} = \frac{e\Phi}{\hbar c}

$$

where $\Phi$ is the magnetic flux through the solenoid.

This phase is:
- **Gauge-invariant** for closed paths (absolute, not depending on choice of gauge)
- **Observable** in double-slit interference: creates shift in fringe pattern
- **Purely quantum**: no classical explanation (classical electron is unaffected by field it never enters)
- **Topological**: depends only on total flux, not on field distribution

The AB effect is a manifestation of Berry phase in position space, unifying electromagnetism with geometric quantum mechanics.

:::

### Quantum Hall Effect and Chern Numbers

When electrons move in a 2D crystal under a strong perpendicular magnetic field, they form **Landau levels**. In the absence of disorder, the Hall conductance is quantized:

$$

\sigma_{xy} = \frac{e^2}{h}\nu

$$

where $\nu$ is an integer (filling factor).

In real crystalline systems without disorder, the conductance can be written as:

$$

\sigma_{xy} = \frac{e^2}{h}\sum_{\text{occupied bands}} C_n

$$

where $C_n$ is the **Chern number** of band $n$:

$$

C_n = \frac{1}{2\pi}\int_{\text{BZ}} \mathcal{F}_{xy}^{(n)}(\boldsymbol{k})\,\mathrm{d}^2k \in \mathbb{Z}

$$

Here $\mathcal{F}_{xy}^{(n)}$ is the Berry curvature of band $n$ in crystal momentum space $\boldsymbol{k}$. The integer quantization is **topological** — it cannot change without closing the band gap.

This is the **TKNN formula** (Thouless, Kohmoto, Nightingale, Nijs, 1982), linking Berry's geometric phase to observable transport phenomena.

:::{admonition} TKNN Formula: Quantum Hall Conductance and Chern Numbers
:class: important

The integer quantum Hall effect is fundamentally a topological phenomenon. The quantized Hall conductivity is:

$$

\sigma_{xy} = \frac{e^2}{h} C_{\text{total}}

$$

where the **Chern number** of each occupied band is:

$$

C_n = \frac{1}{2\pi}\int_{\text{Brillouin Zone}} \mathrm{d}^2k \, \mathcal{F}_{n,xy}(\boldsymbol{k})

$$

with Berry curvature:

$$

\mathcal{F}_{n,xy}(\boldsymbol{k}) = -2\,\mathrm{Im}\sum_{m\neq n}\frac{\langle u_n|\partial_{k_x}\hat{H}|u_m\rangle\langle u_m|\partial_{k_y}\hat{H}|u_n\rangle}{(E_m - E_n)^2}

$$

**Key properties**:
- $C_n$ is always an **integer** (topological invariant)
- Cannot change without closing the band gap
- Each filled Landau level contributes $C = 1$
- Edge states = topologically protected conducting channels

:::

:::{admonition} Berry Curvature in Momentum Space
:class: important

For a Bloch electron with band dispersion $E_n(\boldsymbol{k})$ and Bloch states $|u_n(\boldsymbol{k})\rangle$, the Berry connection in $\boldsymbol{k}$-space is:

$$

\mathcal{A}_{n,\mu}(\boldsymbol{k}) = \mathrm{i}\langle u_n(\boldsymbol{k})|\partial_{k_\mu}u_n(\boldsymbol{k})\rangle

$$

The Berry curvature is:

$$

\mathcal{F}_{n,\mu\nu}(\boldsymbol{k}) = \partial_{k_\mu}\mathcal{A}_{n,\nu} - \partial_{k_\nu}\mathcal{A}_{n,\mu} - 2\,\mathrm{Im}\sum_{m\neq n}\frac{\langle u_n|\partial_{k_\mu}\hat{H}|u_m\rangle\langle u_m|\partial_{k_\nu}\hat{H}|u_n\rangle}{(E_m - E_n)^2}

$$

For a 2D system:

$$

\mathcal{F}_{n,xy}(\boldsymbol{k}) = -2\,\mathrm{Im}\sum_{m\neq n}\frac{\langle u_n|\partial_{k_x}\hat{H}|u_m\rangle\langle u_m|\partial_{k_y}\hat{H}|u_n\rangle}{(E_m - E_n)^2}

$$

:::

### Weak Localization and Antilocalization

In mesoscopic systems (submicron conductors), quantum interference affects transport. Berry phase from spin-orbit coupling modifies these interference patterns:

- **Without spin-orbit**: Constructive interference of time-reversed paths leads to **weak localization** (negative magnetoconductance).
- **With strong spin-orbit**: Berry phase of $\pi$ from a closed loop causes destructive interference, leading to **antilocalization** (positive magnetoconductance).

The transition between these regimes is a direct probe of Berry phase effects in condensed matter.

### Berry Phase in NMR and Magnetic Resonance

In a rotating reference frame, nuclear spins experience an effective magnetic field. As the frame rotates, spins adiabatically follow the field direction. The Berry phase acquired modifies **NMR line shifts** and affects the observed spectrum.

In particular, for coupled spins in a multidimensional NMR experiment, Berry phases from adiabatic pulses accumulate and can be detected as phase shifts in spectral peaks. This is used in sophisticated pulse sequences for structure determination.

:::{admonition} Example: Rotation of Spin-1/2
:class: example

A spin initially in state $|+z\rangle$ experiences an RF field rotating in the $xy$-plane. The effective field precesses about the spin as the RF phase changes. If the RF amplitude is large (adiabatic), the spin stays aligned with the instantaneous effective field.

After one complete RF phase cycle (the field returns to its original state), the spin acquires a Berry phase:

$$

\gamma = -\Omega/2

$$

where $\Omega$ is the solid angle swept on the Bloch sphere. This shifts the NMR frequency by an amount proportional to the spin projection in the effective field direction.

:::

### Pancharatnam Phase in Polarized Light

Light polarization can be represented as a point on the **Poincaré sphere** (a 2-sphere where each point encodes the two components of the polarization state).

When light passes through a sequence of polarization optics (retarders, rotators), its polarization state evolves around a path on the Poincaré sphere. At the end, if the state returns to its starting point (closed path), the light acquires a **Pancharatnam phase**:

$$

\Delta\phi_P = \frac{\Omega}{2}

$$

where $\Omega$ is the solid angle enclosed by the path.

This is a manifestation of Berry phase for photons. It is observable as a measurable phase shift in interference experiments and is used in optical devices and quantum metrology.

:::{admonition} Geometric Phase in Different Physical Systems
:class: important

Berry phase (or geometric phase) appears universally across quantum systems, always related to the solid angle or curvature in an appropriate parameter space:

| System | Parameter Space | Phase | Observable |
|--------|-----------------|-------|-----------|
| Spin-1/2 in rotating $\boldsymbol{B}$ | Direction $(\theta, \phi)$ on unit sphere | $-\Omega/2$ | Interference patterns |
| Molecules (BO approx) | Nuclear coordinates $\boldsymbol{R}$ | $\oint \mathcal{A} \cdot \mathrm{d}\boldsymbol{R}$ | Reaction branching |
| Aharonov-Bohm effect | Electron position around flux | $e\Phi/(\hbar c)$ | Fringe shifts |
| Photon polarization | Poincaré sphere | $+\Omega/2$ | Polarization rotation |
| Bloch electrons in crystals | Crystal momentum $\boldsymbol{k}$ | Chern number | Hall conductance |

The universal feature: **closed paths acquire gauge-invariant phases that depend only on geometry**.
