## Lecture Notes

In superconductors, quantum effects conspire to quantize magnetic flux in units of h/2e (not h/e), a spectacular phenomenon arising from Cooper pairs—composite bosons of two electrons. This reveals a principle deeper than ordinary gauge invariance: when multiple particles form a condensate, topological properties of the wavefunction become fixed, and only certain flux values are allowed. Flux quantization is a macroscopic quantum effect bridging gauge theory, topology, and superconductivity.
:::{admonition} Cooper Pairs and Half-Integer Flux Quantization
:class: important

Superconductors show flux quantization in half-integer units ($\Phi_0 = h/2e$) because Cooper pairs carry charge $2e$. This factor of 2 is a smoking-gun signature of pairing in superconductivity.
:::


### Single-Valued Wavefunction Requirement

A fundamental quantum requirement is that the wavefunction must be **single-valued**: after traveling around any closed loop, the particle returns to the same quantum state. In a superconductor with magnetic flux threaded through a ring, this constraint couples directly to the electromagnetic field.

As a particle (or Cooper pair) travels once around the loop, it accumulates phase from the vector potential:

$$

\Delta\phi = \frac{q}{\hbar c}\oint \boldsymbol{A} \cdot d\boldsymbol{l} = \frac{q\Phi}{\hbar c}

$$

where $\Phi$ is the total magnetic flux enclosed by the loop. For the wavefunction to be single-valued after one complete circuit, the phase must return to its original value (modulo $2\pi$).

:::{admonition} Derivation: Single-Valued Condition
:class: dropdown note


If the wavefunction at angle $\theta = 0$ is $\psi(0)$, then after traveling once around the loop to $\theta = 2\pi$, it becomes:

$$

\psi(2\pi) = e^{i\Delta\phi} \psi(0)

$$

For single-valuedness, we require:

$$

\psi(2\pi) = \psi(0)

$$

This means:

$$

e^{i\Delta\phi} = 1

$$

which requires $\Delta\phi = 2\pi n$ for integer $n$.

:::

### Flux Quantization Condition

Combining the single-valuedness condition with the phase expression gives the **flux quantization** condition:

$$

\frac{q\Phi}{\hbar c} = 2\pi n

$$

Solving for the flux:

$$

\Phi = n\frac{2\pi\hbar c}{q}

$$

**Flux is quantized in discrete units.** We define the **flux quantum** for a particle of charge $q$:

$$

\Phi_0 = \frac{2\pi\hbar c}{q}

$$

Then the allowed fluxes are:

$$

\Phi = n \Phi_0, \quad n = 0, \pm 1, \pm 2, \ldots

$$

:::{admonition} Derivation: Flux Quantum Definition
:class: dropdown note


The flux quantum is a fundamental constant with dimensions of magnetic flux. In SI units with $c$ in m/s:

$$

\hbar = 1.055 \times 10^{-34}\text{ J}\cdot\text{s}, \quad c = 3 \times 10^8 \text{ m/s}

$$

For a charge $q$, the flux quantum becomes:

$$

\Phi_0 = \frac{2\pi\hbar c}{q}

$$

Alternatively, using the relation $2\pi\hbar c = hc$ (since $h = 2\pi\hbar$):

$$

\Phi_0 = \frac{hc}{q}

$$

:::

### Flux Quantum for Different Particles

The flux quantum has different values depending on the charge of the particle involved.

**For electrons** ($q = e$):

$$

\Phi_{0,e} = \frac{h}{e} \approx 4.14 \times 10^{-15}\text{ Wb}

$$

**For Cooper pairs** in superconductors ($q = 2e$, since pairs carry twice the electron charge):

$$

\Phi_{0,\text{pair}} = \frac{h}{2e} = \frac{\Phi_{0,e}}{2} \approx 2.07 \times 10^{-15}\text{ Wb}

$$

**This is the key difference**: Cooper pairs carry twice the electron charge, so their flux quantum is **exactly half** the single-electron flux quantum. This fundamental distinction explains why superconductors exhibit unique quantum phenomena unavailable to normal metals.

:::{admonition} Derivation: Why Cooper Pair Charge is 2e
:class: dropdown note


In a superconductor, electrons near the Fermi surface form **Cooper pairs**—composite bosons bound by the attractive electron-phonon interaction (BCS theory). A Cooper pair consists of two electrons:

$$

\text{Cooper pair} = \text{electron 1} + \text{electron 2}

$$

The total charge is therefore $q_{\text{pair}} = (-e) + (-e) = -2e$, or $|q_{\text{pair}}| = 2e$.

This doubling of charge leads directly to halving of the flux quantum:

$$

\Phi_{0,\text{pair}} = \frac{h}{2e} = \frac{\Phi_{0,e}}{2}

$$

:::

### Persistent Currents in Superconducting Rings

When magnetic flux threads a superconducting ring, a remarkable phenomenon emerges: **persistent currents** that flow indefinitely without dissipation. The superconductor cannot change the flux through the ring (due to flux quantization), so it responds by generating a current that maintains the current quantum state.

When external magnetic field tries to change the flux, the superconductor generates a **Meissner current**:

$$

\boldsymbol{B}_{\text{applied}} + \boldsymbol{B}_{\text{induced}} = 0 \quad (\text{Meissner effect})

$$

This current persists indefinitely—**no dissipation, no decay**—because any attempt to change the flux would violate flux quantization. The quantum state of the superconducting condensate (characterized by the phase of its macroscopic wavefunction) gets "locked" to the flux through the ring.

:::{admonition} Derivation: Phase Locking and Persistent Current
:class: dropdown note


The superconducting condensate has a macroscopic wavefunction:

$$

\psi_{\text{cond}}(\boldsymbol{r}) = |\psi_0(\boldsymbol{r})|e^{i\theta(\boldsymbol{r})}

$$

As we traverse the superconducting ring, the phase winds by exactly $2\pi$ times the flux quantum number:

$$

\oint \nabla\theta \cdot d\boldsymbol{l} = 2\pi n

$$

The quantum state is "trapped" with fixed winding number $n$. If external flux tries to increase, the superconductor immediately generates a screening current to maintain constant $n$. This current produces the equal and opposite magnetic field (Meissner effect).

Since the superconductor has zero electrical resistance, this current experiences no ohmic dissipation and persists indefinitely.

:::

### Energy and Persistent Current

The energy cost of changing the flux through a superconducting loop is enormous because it requires creating quasiparticle excitations (breaking Cooper pairs). Below the superconducting gap, this cost is exponentially large, making flux change practically impossible. The persistent current can be thought of as the superconductor's response to maintain constant flux despite external perturbations.

### SQUID: Superconducting Quantum Interference Device

The **SQUID** (Superconducting Quantum Interference Device) is one of the most sensitive magnetometers ever invented, exploiting Josephson junctions and flux quantization to measure magnetic fields. A SQUID consists of a superconducting loop interrupted by one or two **Josephson junctions**—thin insulating barriers through which Cooper pairs can tunnel.

**Operating principle**: The phase around the superconducting loop must satisfy flux quantization:

$$

\Delta\phi_{\text{loop}} = \frac{2e\Phi}{\hbar c} = 2\pi n + \delta

$$

where $\Phi$ is the total enclosed flux and $\delta$ is the phase slip at the junctions. The **critical current** $I_c$ of the junction (maximum supercurrent before switching to resistive state) depends sensitively on $\delta$, and thus on the enclosed flux:

$$

I_c(\Phi) = I_{c0}\left|\cos\left(\frac{\pi\Phi}{\Phi_{0,\text{pair}}}\right)\right|

$$

This oscillation with period $\Phi_{0,\text{pair}} \approx 2 \times 10^{-15}$ Wb makes SQUIDs extraordinarily sensitive—they can detect flux changes as small as a fraction of one flux quantum, equivalent to magnetic field variations of ~$10^{-18}$ Tesla. This makes SQUIDs the most sensitive magnetometers known, used in medical imaging (magnetoencephalography) and fundamental physics experiments.

:::{admonition} Derivation: SQUID Critical Current Oscillation
:class: dropdown note


A Josephson junction's supercurrent depends on the phase difference across it:

$$

I = I_{c0} \sin(\phi_L - \phi_R)

$$

where $\phi_L$ and $\phi_R$ are the phases on left and right sides.

For a single-junction SQUID (or analyzing one path through two junctions), the phase difference depends on the flux:

$$

\phi_L - \phi_R = \frac{2e\Phi}{\hbar c} + \text{const}

$$

The maximum current (critical current) occurs when $\sin(\phi_L - \phi_R) = \pm 1$, giving:

$$

I_c(\Phi) \propto \left|\cos\left(\frac{\pi\Phi}{\Phi_{0,\text{pair}}}\right)\right|

$$

This oscillates with period equal to one flux quantum. Measuring $I_c$ as a function of external magnetic field reveals the periodic modulation—a direct signature of flux quantization.

:::

### Topological Quantization

Flux quantization is fundamentally **topological**—it depends only on the winding number of the phase around the loop, not on microscopic details:

- Works for any superconductor (conventional, heavy-fermion, high-$T_c$)
- Insensitive to material properties, disorder, or defects
- Holds as long as the superconducting gap exists

This topological robustness is what makes SQUIDs so powerful—their operation doesn't depend sensitively on microscopic parameters, making them extraordinarily reliable and precise.

### Summary

- **Single-valuedness**: Wavefunctions must return to themselves after encircling flux: $\psi(2\pi) = \psi(0)$.
- **Flux quantization**: Enclosed flux must satisfy $\Phi = n\Phi_0$ with integer $n$; a topological quantization condition.
- **Flux quantum**: $\Phi_0 = h/q$ for charge $q$; for Cooper pairs ($q=2e$): $\Phi_0 = h/2e \approx 2.07 \times 10^{-15}$ Wb.
- **Persistent currents**: Superconductors maintain dissipation-free currents that lock the phase to enclosed flux.
- **Meissner effect**: Superconductors expel magnetic fields (except quantized flux units); zero field inside despite external fields.
- **SQUIDs**: Josephson junction pairs create the world's most sensitive magnetometers via interference modulation.
- **Topological origin**: Flux quantization doesn't depend on microscopic details—only on winding of quantum phase.
- **Fundamental verification**: Superconducting experiments provide direct evidence for quantum mechanics at macroscopic scales.
