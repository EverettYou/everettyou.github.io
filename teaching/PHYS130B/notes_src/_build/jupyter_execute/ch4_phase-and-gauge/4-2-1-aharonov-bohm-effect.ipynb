## Lecture Notes

The Aharonov-Bohm (AB) effect is one of the most celebrated phenomena in quantum mechanics. It demonstrates that the quantum phase of a particle can be affected by a vector potential even in regions where the electromagnetic field (the field strength tensor) vanishes completely. This effect reveals the fundamental importance of gauge fields in quantum mechanics and hints at deep geometric structure underlying quantum theory.

### The Aharonov-Bohm Experimental Setup

#### Physical Configuration

Imagine a solenoid (a coil wound tightly on a cylinder) producing a uniform magnetic field $\boldsymbol{B}$ inside the solenoid but approximately zero outside. An electron beam is split into two paths that encircle the solenoid—one path passing above it and one passing below. The two beams then recombine at a distant observation screen.

In classical mechanics, since the electron experiences no Lorentz force ($\boldsymbol{F} = q\boldsymbol{v} \times \boldsymbol{B}$) in the field-free region outside the solenoid, both paths would follow identical trajectories. The interference pattern at the screen should show no dependence on the solenoid's magnetic field.

In quantum mechanics, something startling happens:

:::{admonition} Aharonov-Bohm Phase Shift
:class: important

Even though the magnetic field is zero along the electron's path, the quantum phase of the electron wavefunction *changes* depending on the solenoid's flux. This phase shift shifts the interference pattern—a purely quantum effect with no classical analog.

The phase difference between the upper and lower paths is:

$$


\Delta\phi = \frac{q\Phi_B}{\hbar c}

$$


where $\Phi_B = \int \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S}$ is the total magnetic flux *through* the solenoid (not experienced by the particle!).
:::

### Vector Potential in a Solenoid

To understand the AB effect, we must think in terms of the vector potential $\boldsymbol{A}$, not the field strength $\boldsymbol{B}$.

#### Outside the Solenoid

For a solenoid of radius $R$ with uniform internal field $B$ along the $z$-axis, the vector potential outside (in cylindrical coordinates) is:

$$


\boldsymbol{A}(r > R) = \frac{\Phi_B}{2\pi r}\hat{\boldsymbol{\theta}}

$$


where $\Phi_B = \pi R^2 B$ is the total flux. Note:
- $\boldsymbol{A}$ is *nonzero* outside the solenoid
- The magnetic field $\boldsymbol{B} = \nabla \times \boldsymbol{A} = 0$ outside (the curl of $\boldsymbol{A}$ vanishes)

#### Circulation Integral

The key property is that the vector potential's circulation around any closed loop encircling the solenoid equals the enclosed flux:

$$


\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \Phi_B

$$


This result holds for *any* closed path enclosing the solenoid, regardless of its shape. It is a topological fact—a consequence of Stokes' theorem:

$$


\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \int_{\text{surface}} (\nabla \times \boldsymbol{A}) \cdot \mathrm{d}\boldsymbol{S} = \int_{\text{surface}} \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S} = \Phi_B

$$


:::{admonition} Discussion
:class: dropdown tip

The vector potential $\boldsymbol{A}$ is nonzero outside the solenoid, but the magnetic field $\boldsymbol{B} = \nabla \times \boldsymbol{A} = 0$. In classical mechanics, only $\boldsymbol{B}$ couples to a moving charge. How can the nonzero $\boldsymbol{A}$ produce a physical effect (phase shift) if no classical force acts? Does this mean quantum mechanics sees something in the vector potential that classical mechanics misses?
:::

### Phase Accumulation Along a Path

#### The Phase Integral

When a charged particle travels along a path $\mathcal{C}$ through a region with vector potential $\boldsymbol{A}$, the quantum phase acquired is:

$$


\phi[\mathcal{C}] = \frac{q}{\hbar c}\int_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}

$$


This arises from the covariant derivative in the Schrödinger equation: as the wavefunction propagates, the local vector potential "rotates" its phase continuously.

#### Two-Path Interference

Consider two paths: the upper path (encircling the solenoid from above) and the lower path (from below). Both start and end at the same points, but their relative phase depends on the solenoid's flux.

The phase difference is:

$$


\Delta\phi = \frac{q}{\hbar c}\left(\int_{\text{upper}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} - \int_{\text{lower}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}\right) = \frac{q}{\hbar c}\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q\Phi_B}{\hbar c}

$$


The two wave packets acquire a relative phase that manifests as a shift in the interference pattern.

### Gauge Invariance of the Aharonov-Bohm Effect

A crucial check: is the AB effect gauge-invariant?

:::{admonition} Gauge Invariance of AB Phase
:class: important

Under a gauge transformation $\boldsymbol{A} \to \boldsymbol{A}' = \boldsymbol{A} - \nabla\chi$ and $\psi \to \psi' = \mathrm{e}^{\mathrm{i}q\chi}\psi$:

The accumulated phase changes as:

$$


\phi'[\mathcal{C}] = \frac{q}{\hbar c}\int_{\mathcal{C}} (\boldsymbol{A} - \nabla\chi) \cdot \mathrm{d}\boldsymbol{l} = \phi[\mathcal{C}] - \frac{q}{\hbar c}[\chi(\text{end}) - \chi(\text{start})]

$$


But the wavefunction also acquires an overall phase factor $\mathrm{e}^{\mathrm{i}q(\chi_{\text{end}} - \chi_{\text{start}})}$. These *precisely cancel* when we compare *relative phases* of two paths with the same endpoints:

$$


\Delta\phi' = \phi'_1 - \phi'_2 = (\phi_1 - \Delta\chi_1) - (\phi_2 - \Delta\chi_2) = \phi_1 - \phi_2 = \Delta\phi

$$


The **relative phase difference between two interfering beams is gauge-invariant**.
:::

:::{admonition} Discussion
:class: dropdown tip

We showed that the relative phase difference $\Delta\phi' = \Delta\phi$ is gauge-invariant. But the individual phases $\phi_1'$ and $\phi_2'$ on each path *do* change under a gauge transformation. How can we experimentally measure the relative phase in a way that reveals the gauge-invariant quantity, not the gauge-dependent individual phases?
:::

### Why This Matters: The Non-Classical Nature of the Vector Potential

The Aharonov-Bohm effect reveals something profound:

:::{admonition} The Vector Potential is Physical
:class: note

In classical electromagnetism, the vector potential $\boldsymbol{A}$ is regarded as a mathematical convenience—only the field strengths $\boldsymbol{E}$ and $\boldsymbol{B}$ have direct physical meaning.

In quantum mechanics, $\boldsymbol{A}$ appears *explicitly* in the Schrödinger equation via the covariant derivative. The AB effect demonstrates that $\boldsymbol{A}$ produces observable consequences even when $\boldsymbol{B} = 0$ everywhere on the particle's path.

This is not a contradiction: $\boldsymbol{A}$ is gauge-dependent (unobservable in isolation), but the *circulation* $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$ is gauge-invariant (observable). Only gauge-invariant combinations appear in interference patterns.
:::

### Topological Nature and Flux Quantization

The AB effect has a topological character: it depends only on the total flux enclosed, not on the detailed field configuration.

:::{admonition} Topology: Shape-Independence
:class: dropdown

Imagine deforming the two paths continuously without breaking them. As long as they remain separated and encircle the solenoid, the phase difference $\Delta\phi = q\Phi_B/(\hbar c)$ remains unchanged. This is a topological property: it is invariant under continuous deformation.

Only discontinuous changes (like one path "jumping over" the solenoid) would change the phase.
:::

This topological structure connects to **flux quantization** in superconductivity: when a superconductor forms a ring, quantum interference enforces that the flux through the ring is quantized in units of the flux quantum:

$$


\Phi_q = \frac{\pi \hbar c}{e} \approx 4.1 \times 10^{-15} \text{ Wb}

$$


When the enclosed flux is exactly $\Phi_q$, the phase shift is $2\pi$, returning the electron wavefunction to itself with a complete cycle.

### Summary

- The **Aharonov-Bohm effect** demonstrates that quantum phases depend on the vector potential $\boldsymbol{A}$, even in regions where the field strength (magnetic field) is zero.
- The **phase shift** around a closed path equals $q\Phi_B/(\hbar c)$, where $\Phi_B$ is the enclosed flux.
- The effect is **gauge-invariant**: relative phases between interfering beams are independent of gauge choice.
- The AB effect reveals that the vector potential has genuine physical significance in quantum mechanics, not merely mathematical convenience as in classical theory.
- The topological nature of the AB effect connects to **flux quantization** and appears in superconductivity, quantum Hall effect, and modern topological materials.
