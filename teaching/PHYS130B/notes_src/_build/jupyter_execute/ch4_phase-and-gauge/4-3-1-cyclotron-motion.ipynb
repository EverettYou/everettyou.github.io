## Lecture Notes

A charged particle in a magnetic field orbits in a circle at the cyclotron frequency—a phenomenon that seems classically mundane but is profoundly quantum. The cyclotron frequency ωc = eB/m is universal: it depends only on the physical field, not on gauge choice, and is independent of the particle's energy. This classical foundation sets the stage for understanding Landau quantization, the quantum Hall effect, and the intimate connection between geometry (curvature of orbits) and quantum mechanics (quantization of energy levels).
:::{admonition} Why Cyclotron Frequency is Gauge-Independent
:class: important

The cyclotron frequency $\omega_c = eB/m$ depends only on the physical magnetic field **B**, not on the gauge choice for the vector potential. This is your first clue that observables are gauge-invariant.
:::


### Lorentz Force and Equations of Motion

A charged particle of charge $q = -e$ (electron) and mass $m$ moving in a magnetic field $\boldsymbol{B}$ experiences the **Lorentz force**:

$$

\boldsymbol{F} = q(\boldsymbol{v} \times \boldsymbol{B})

$$

This leads to Newton's second law:

$$

m\frac{d\boldsymbol{v}}{dt} = q(\boldsymbol{v} \times \boldsymbol{B})

$$

For a uniform field $\boldsymbol{B} = B\hat{z}$, the motion parallel to the field is unaffected, while the perpendicular components couple to the field and cause the particle to orbit.

:::{admonition} Derivation: Component Equations
:class: dropdown note


Writing out components for $\boldsymbol{B} = B\hat{z}$:

$$

m\frac{dv_x}{dt} = qv_y B, \quad m\frac{dv_y}{dt} = -qv_x B, \quad m\frac{dv_z}{dt} = 0

$$

Motion in the $z$-direction is free motion: $v_z = \text{const}$. The $x$-$y$ components are coupled and oscillate.

:::

### Cyclotron Frequency

The perpendicular motion is characterized by a universal frequency—the **cyclotron frequency**:

$$

\omega_c = \frac{|q|B}{m} = \frac{eB}{m}

$$

This is a fundamental frequency: it depends only on the field strength $B$ and the particle's charge-to-mass ratio, **not on the particle's energy or speed**. All particles with the same charge-to-mass ratio orbit at the same frequency, regardless of how fast they move.

The perpendicular equations of motion become:

$$

\frac{dv_x}{dt} = \omega_c v_y, \quad \frac{dv_y}{dt} = -\omega_c v_x

$$

These are **harmonic oscillator equations**: the velocity rotates in the perpendicular plane with angular frequency $\omega_c$.

:::{admonition} Derivation: Solving the Harmonic Equations
:class: dropdown note


Taking the time derivative of the first equation and substituting the second:

$$

\frac{d^2v_x}{dt^2} = \omega_c\frac{dv_y}{dt} = -\omega_c^2 v_x

$$

This is simple harmonic motion. The general solution is:

$$

v_x(t) = v_\perp\cos(\omega_c t + \phi), \quad v_y(t) = v_\perp\sin(\omega_c t + \phi)

$$

where $v_\perp = \sqrt{v_x^2 + v_y^2}$ is the perpendicular speed and $\phi$ is a phase determined by initial conditions.

:::

### Circular Orbits

Integrating the velocity gives the trajectory: the particle traces a **perfect circle** in the plane perpendicular to the magnetic field. The orbit has radius:

$$

r_c = \frac{v_\perp}{\omega_c} = \frac{mv_\perp}{eB}

$$

This is the **cyclotron radius** (also called **Larmor radius**)—it depends on the particle's perpendicular speed. Faster particles orbit in larger circles. Meanwhile, the particle drifts freely along the field direction:

$$

z(t) = z_0 + v_z t

$$

In classical mechanics, any perpendicular kinetic energy is allowed, so the cyclotron radius can be made arbitrarily large or small. **This is fundamentally different from quantum mechanics**, where energy is quantized into discrete Landau levels and the radius becomes a quantum property.

:::{admonition} Derivation: Trajectory Integration
:class: dropdown note


Starting from the velocity solution with $\phi = 0$:

$$

x(t) = x_c + r_c\sin(\omega_c t), \quad y(t) = y_c + r_c\cos(\omega_c t)

$$

where $(x_c, y_c)$ is the **guiding center** (center of the orbit). The particle traces a circle of radius $r_c$ in the $x$-$y$ plane while moving freely in the $z$-direction.

The centripetal acceleration $a = v_\perp\omega_c$ is provided by the Lorentz force $F = ev_\perp B$, consistent with $F = ma$.

:::

### Kinetic Energy and Orbital Period

The perpendicular kinetic energy is $E_\perp = \frac{1}{2}mv_\perp^2$, which determines the cyclotron radius.

The orbital period is:

$$

T_c = \frac{2\pi}{\omega_c} = \frac{2\pi m}{eB}

$$

**Remarkably, the period is independent of energy**: slow and fast particles orbit at the same frequency. This is a special feature of the cyclotron motion and underlies the cyclotron resonance used to accelerate particles in cyclotrons.

### Hall Effect: Classical Derivation

Consider electrons drifting through a conductor in a current, subject to a perpendicular magnetic field. The Lorentz force deflects electrons to one side of the conductor, creating a **Hall electric field** that opposes further deflection. At equilibrium:

$$

E_H = v_d B

$$

where $v_d$ is the drift velocity. This Hall field is **transverse to the current**—perpendicular to both the current direction and the magnetic field. This is the **Hall effect**: a voltage develops perpendicular to both current and field, revealing information about the charge carriers.

:::{admonition} Derivation: Hall Voltage and Conductance
:class: dropdown note


For a rectangular conductor with current $I$ in the $x$-direction and field $B$ in the $-z$ direction, electrons (charge $-e$) drift with velocity $\boldsymbol{v} = v_d\hat{x}$.

The Lorentz force is:

$$

\boldsymbol{F} = -e(\boldsymbol{v} \times \boldsymbol{B}) = -ev_d B\hat{y}

$$

Electrons accumulate on one side, creating a Hall electric field $E_y$ in the $y$-direction. At equilibrium:

$$

eE_y = ev_d B \quad \Rightarrow \quad E_y = v_d B

$$

The Hall voltage across width $w$ is:

$$

V_H = E_y \cdot w = v_d B w

$$

The Hall conductance (or Hall resistance) depends on the charge carrier density, revealing whether carriers are electrons, holes, or something else.

:::

### Classical Hall Angle

In the presence of both magnetic and electric fields with collisions, the drift velocity is determined by the **Hall angle** $\theta_H$:

$$

\tan\theta_H = \frac{\omega_c\tau}{1}

$$

where $\tau$ is the collision time. For weak fields ($\omega_c\tau \ll 1$), the Hall effect is linear in $B$. For strong fields ($\omega_c\tau \gg 1$), particles complete many orbits between collisions, and quantum effects dominate—the **quantum Hall regime**.

### Summary

- **Lorentz force**: $\boldsymbol{F} = q(\boldsymbol{v} \times \boldsymbol{B})$ causes perpendicular deflection.
- **Cyclotron frequency**: $\omega_c = eB/m$ is the orbital frequency, independent of particle energy.
- **Cyclotron radius**: $r_c = mv_\perp/(eB)$ increases with perpendicular velocity.
- **Orbital period**: $T = 2\pi m/(eB)$ is independent of velocity—universal for a given $B$ and particle type.
- **Drift motion**: Parallel motion is unaffected; particles drift freely along field lines.
- **Guiding center**: The center of the circular orbit slowly drifts, decoupling fast (cyclotron) from slow (drift) motion.
- **Hall effect**: Transverse electric field develops from Lorentz force on current-carrying charges.
- **Hall coefficient**: $R_H = 1/(ne)$ provides direct measurement of charge carrier density.
- **Classical limitation**: Any energy is allowed; quantization appears only in quantum mechanics.
