## Lecture Notes

The motion of a charged particle in a uniform magnetic field is one of the most fundamental problems in electromagnetism and quantum mechanics. It serves as the foundation for understanding the quantum Hall effect, shows how Landau quantization arises, and reveals deep connections between classical and quantum physics.

### Classical Cyclotron Motion

#### Equations of Motion

A charged particle with charge $q$ and mass $m$ in a uniform magnetic field $\boldsymbol{B} = B\hat{z}$ experiences the Lorentz force:

$$

m\boldsymbol{a} = q\boldsymbol{v} \times \boldsymbol{B}

$$

Writing $\boldsymbol{v} = v_x\hat{x} + v_y\hat{y} + v_z\hat{z}$ and $\boldsymbol{a} = \dot{v}_x\hat{x} + \dot{v}_y\hat{y} + \dot{v}_z\hat{z}$, the cross product gives:

$$

m\dot{v}_x = qv_yB, \quad m\dot{v}_y = -qv_xB, \quad m\dot{v}_z = 0

$$

:::{admonition} Cyclotron Frequency
:class: important

The characteristic timescale of motion perpendicular to $\boldsymbol{B}$ is set by the **cyclotron frequency** (or **Larmor frequency**):

$$

\omega_c = \frac{qB}{m}

$$

(In Gaussian units: $\omega_c = eB/mc$.)

**Key properties**:
- Independent of particle kinetic energy
- Depends only on charge-to-mass ratio and field strength
- Determines the period of one orbit: $T_c = 2\pi/\omega_c$
- For electrons: $\omega_c = eB/m_e$ (negative because $e < 0$)

:::

The equations become:

$$

\ddot{v}_x + \omega_c^2 v_x = 0, \quad \ddot{v}_y + \omega_c^2 v_y = 0

$$

These are simple harmonic oscillators. The general solution is:

$$

v_x(t) = v_\perp \cos(\omega_c t + \phi_0), \quad v_y(t) = v_\perp \sin(\omega_c t + \phi_0)

$$

where $v_\perp = \sqrt{v_x^2 + v_y^2}$ is the speed perpendicular to $\boldsymbol{B}$ and $\phi_0$ is an arbitrary phase.

#### Circular Orbits and Guiding Center Motion

Integrating the velocity components gives the orbit in the $xy$-plane:

$$

x(t) = x_0 + \frac{v_\perp}{\omega_c}\sin(\omega_c t + \phi_0), \quad y(t) = y_0 - \frac{v_\perp}{\omega_c}\cos(\omega_c t + \phi_0)

$$

This describes a circular orbit with:

:::{admonition} Cyclotron Radius (Larmor Radius)
:class: important

$$

r_c = \frac{v_\perp}{\omega_c} = \frac{mv_\perp}{qB}

$$

**Physical meaning**:
- The particle spirals around the **guiding center** position $(x_0, y_0)$ at frequency $\omega_c$
- Motion parallel to $\boldsymbol{B}$ (the $z$-direction) is unaffected
- Radius scales with: momentum ($\propto v_\perp$), inversely with field ($\propto 1/B$)
- Different particles with same kinetic energy orbit with different radii if they have different $m/q$ ratios

**Relation to energy**: Using $E_\perp = \frac{1}{2}mv_\perp^2$:

$$

r_c = \frac{\sqrt{2mE_\perp}}{qB}

$$

:::

:::{admonition} Orbital Period: Energy-Independent
:class: important

The period of one complete orbit is:

$$

T_c = \frac{2\pi}{\omega_c} = \frac{2\pi m}{qB}

$$

**Remarkable fact**: $T_c$ is **completely independent of the particle's kinetic energy** or velocity! 

This is why cyclotron accelerators work: particles always take the same time to complete a semicircle, allowing constant-frequency RF acceleration to keep all particles synchronized regardless of their energy.

:::

#### Motion in Crossed $\boldsymbol{E}$ and $\boldsymbol{B}$ Fields

When both electric and magnetic fields are present, the behavior is richer. If $\boldsymbol{E}$ is perpendicular to $\boldsymbol{B}$, the guiding center drifts with velocity:

$$

\boldsymbol{v}_{\text{drift}} = \frac{\boldsymbol{E} \times \boldsymbol{B}}{B^2}

$$

:::{admonition} E×B Drift Velocity
:class: important

In crossed fields $\boldsymbol{E} \perp \boldsymbol{B}$, the guiding center drifts with velocity:

$$

\boldsymbol{v}_E = \frac{\boldsymbol{E} \times \boldsymbol{B}}{B^2}

$$

**Remarkable properties**:
- **Independent of charge** $q$ and **mass** $m$ — all particles drift together
- **Direction**: perpendicular to both $\boldsymbol{E}$ and $\boldsymbol{B}$
- **Magnitude**: $v_E = E/B$ (for perpendicular fields)
- Acts as an **E-cross-B drift** that is fundamental in plasma physics

This universal drift is crucial for plasma confinement and appears throughout geophysics (ionosphere, magnetosphere) and astrophysics (accretion disks, solar flares).

:::

:::{admonition} Note: Physical Intuition for Drift
:class: note

The electric field accelerates the particle in one direction, increasing the cyclotron radius on one side of the orbit. This asymmetry causes the guiding center to drift perpendicular to both $\boldsymbol{E}$ and $\boldsymbol{B}$. The drift speed adjusts so the net force on the guiding center is zero in the drift direction.
:::

### Semiclassical Quantization and Landau Levels

The classical cyclotron motion suggests a natural quantization condition. The action integral for one complete orbit is:

$$

S = \oint p \, \mathrm{d}q = m v_\perp (2\pi r_c) = \frac{2\pi m v_\perp^2}{\omega_c} = \frac{2\pi E_\perp}{\omega_c}

$$

where $E_\perp = \frac{1}{2}m v_\perp^2$ is the kinetic energy perpendicular to the field.

By the Bohr-Sommerfeld quantization rule, $S = 2\pi n \hbar$ for integer $n$, we get:

:::{admonition} Landau Levels (Semiclassical Derivation)
:class: important

Using the Bohr-Sommerfeld quantization condition $\oint p \, \mathrm{d}q = 2\pi\hbar(n + \tfrac{1}{2})$:

$$

E_n = \hbar\omega_c\left(n + \tfrac{1}{2}\right)

$$

for $n = 0, 1, 2, \ldots$

These are the **Landau levels**, equally spaced in energy with gap $\hbar\omega_c$.

**Key features**:
- Spacing $\hbar\omega_c$ is independent of magnetic field strength (but $\omega_c \propto B$, so levels get closer in real energy units as $B \to 0$)
- Each level is highly degenerate (degeneracy $\sim eB\hbar/(\pi c)$ for large systems)
- Semiclassical result is **exact** for the harmonic oscillator problem, explaining why Bohr-Sommerfeld works perfectly here
- Landau quantization is the foundation for the integer and fractional quantum Hall effects

:::

:::{admonition} Derivation: Bohr-Sommerfeld Result
:class: dropdown

The phase space action for circular motion at fixed energy is:

$$

\oint p_\phi \, \mathrm{d}\phi = p_\phi \cdot 2\pi = m r_c v_\perp \cdot 2\pi = 2\pi m \frac{E_\perp}{\omega_c}

$$

Setting $\oint p \, \mathrm{d}q = 2\pi\hbar(n + \tfrac{1}{2})$ (with the zero-point correction $+\tfrac{1}{2}$ characteristic of harmonic oscillators), we obtain:

$$

2\pi m \frac{E_\perp}{\omega_c} = 2\pi\hbar\left(n + \tfrac{1}{2}\right)

$$

$$

E_\perp = \hbar\omega_c\left(n + \tfrac{1}{2}\right)

$$

The $+\tfrac{1}{2}$ zero-point energy is unavoidable from the uncertainty principle: we cannot have both $p$ and $x$ exactly zero for a confined particle.
:::

### Key Physical Insights

:::{admonition} Summary: Cyclotron Motion
:class: important

**Cyclotron frequency** (energy-independent):

$$

\omega_c = \frac{qB}{m}

$$

**Cyclotron radius** (momentum-dependent):

$$

r_c = \frac{mv_\perp}{qB}

$$

**Orbital period** (energy-independent):

$$

T_c = \frac{2\pi m}{qB}

$$

**E×B drift** (independent of $q$, $m$, energy):

$$

\boldsymbol{v}_E = \frac{\boldsymbol{E} \times \boldsymbol{B}}{B^2}

$$

**Landau levels** (from Bohr-Sommerfeld):

$$

E_n = \hbar\omega_c\left(n + \tfrac{1}{2}\right)

$$

These are the fundamental building blocks for understanding:
- Plasma confinement and dynamics
- Quantum Hall effects
- Cyclotron resonance spectroscopy
- Motion in the Earth's magnetosphere

:::

:::{admonition} Discussion
:class: dropdown warning

The Bohr-Sommerfeld quantization rule predicts energy levels $E_n = (n + \tfrac{1}{2})\hbar\omega_c$, exactly matching the quantum mechanical result from Landau quantization. But is this agreement a coincidence, or is there a deeper reason?

(a) In general, Bohr-Sommerfeld is an approximation valid only for large $n$. Yet for cyclotron motion, it gives the *exact* result even for $n=0$. Why is cyclotron motion special?

(b) The $+\tfrac{1}{2}$ zero-point energy comes from the uncertainty principle: you cannot have both $p$ and $q$ simultaneously zero. Can you show that the zero-point energy $\tfrac{1}{2}\hbar\omega_c$ is unavoidable quantum mechanically, using only $\Delta x \Delta p \gtrsim \hbar$ and the circular orbit geometry?

(c) In the classical limit ($\hbar \to 0$ or large $B$), do the Landau levels remain equally spaced? What happens to the energy gaps between levels?
:::

The classical cyclotron problem is the springboard to understanding quantum many-body physics in strong magnetic fields, where the Landau level degeneracy and topological properties dominate the physics.
