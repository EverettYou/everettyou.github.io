## Lecture Notes

Spin is not a small sphere spinning, nor is it classical angular momentum—it is a purely quantum mechanical intrinsic angular momentum with no macroscopic counterpart. A spin-1/2 system has only two states, yet already exhibits quantum mysteries: superposition, non-commutativity of spin components, and intrinsic magnetic moments that are not generated by charge circulation. This section builds the foundation for understanding how spin couples to electromagnetic fields and defines the geometry of parameter space in gauge theory.
:::{admonition} Angular Momentum is NOT Spin
:class: caution

A spinning classical sphere acquires angular momentum. But quantum **spin** is intrinsic angular momentum with no classical analog—it has no "spinning" interpretation. The $S=1/2$ eigenvalue shows spin is irreducibly quantum.
:::


### Angular Momentum Decomposition

A macroscopic spinning object (like a spinning top or gyroscope) has total angular momentum $\boldsymbol{L}$. We can decompose this into two parts by introducing the spin axis direction $\boldsymbol{n}$ (a unit vector):

$$

\boldsymbol{L} = L_\parallel\boldsymbol{n} + \boldsymbol{L}_\perp

$$

where:
- $L_\parallel = \boldsymbol{L} \cdot \boldsymbol{n}$ is the component of angular momentum **along** the spin axis
- $\boldsymbol{L}_\perp$ is the perpendicular part, **orthogonal** to $\boldsymbol{n}$

The kinetic energy of a spinning rigid body is:

$$

T = \frac{L_\parallel^2}{2I_a} + \frac{L_\perp^2}{2I_t}

$$

where:
- $I_a$ is the moment of inertia about the spin axis (axial, typically smaller)
- $I_t$ is the moment of inertia perpendicular to the spin axis (transverse, typically larger)

For a symmetric top (like a spinning coin), $I_t$ is the same for all directions perpendicular to the axis.


When an external torque $\boldsymbol{\tau}$ acts on the spinning object, the total angular momentum evolves as:

$$

\frac{d\boldsymbol{L}}{dt} = \boldsymbol{\tau}

$$

and the spin axis $\boldsymbol{n}$ follows the equation:

$$

\frac{d\boldsymbol{n}}{dt} = \boldsymbol{n} \times \boldsymbol{\Omega}(\boldsymbol{n})

$$

For a potential torque $\boldsymbol{\tau} = -\nabla V(\boldsymbol{n})$, the effective angular velocity is:

$$

\boldsymbol{\Omega}(\boldsymbol{n}) = \frac{1}{I_t}(\nabla V \times \boldsymbol{n})

$$

where the gradient is taken with respect to the direction on the unit sphere. This structure shows that spin dynamics is governed by an effective field that depends on both the potential energy landscape and the current spin direction.

### Formal Analogy with Lorentz Force

The equation $\frac{d\boldsymbol{n}}{dt} = \boldsymbol{n} \times \boldsymbol{\Omega}$ is **formally identical** to the Lorentz force equation:

$$

m\boldsymbol{a} = q(\boldsymbol{v} \times \boldsymbol{B})

$$

The correspondence is:

| Spin dynamics | Lorentz force |
|---|---|
| Spin axis $\boldsymbol{n}$ | Velocity $\boldsymbol{v}$ |
| Effective field $\boldsymbol{\Omega}$ | Magnetic field $\boldsymbol{B}$ |
| Potential energy $V(\boldsymbol{n})$ | Electric potential $\Phi$ |
| Unit sphere $\|\boldsymbol{n}\| = 1$ | Velocity direction (magnitude) |

This mapping means that **spin dynamics on the unit sphere is equivalent to the motion of a charged particle in a magnetic field**. This is not a coincidence—it reveals a deep geometric structure that will connect to monopoles.

### Precession as Geometrical Motion

In the absence of external torque ($V = 0$), the spin axis precesses with constant angular velocity $\boldsymbol{\Omega}_0$:

$$

\frac{d\boldsymbol{n}}{dt} = \boldsymbol{n} \times \boldsymbol{\Omega}_0

$$

This means the spin axis traces out a circle on the sphere (or cone in 3D space), always maintaining constant angle relative to the precession axis.

For weak external potential $V(\boldsymbol{n})$, perturbation theory shows that the spin axis slowly precesses around the direction of $\nabla V$ with frequency proportional to $|\nabla V|/I_t$.

### Classical Spin Dynamics in a Magnetic Field

When a **magnetic dipole** with moment $\boldsymbol{\mu}$ is placed in an external magnetic field $\boldsymbol{B}$, it experiences a **torque**:

$$

\boldsymbol{\tau} = \boldsymbol{\mu} \times \boldsymbol{B}

$$

For a spinning object with spin angular momentum $\boldsymbol{S}$, the magnetic moment is proportional to the spin:

$$

\boldsymbol{\mu} = -\frac{g}{2m}\boldsymbol{S}

$$

where $g$ is the **g-factor** (unitless, typically $g \approx 2$ for the electron) and $m$ is the mass. The negative sign (by convention) indicates that the magnetic moment opposes the spin angular momentum for particles with negative charge.

The equation of motion for the angular momentum is:

$$

\frac{d\boldsymbol{S}}{dt} = \boldsymbol{\tau} = \boldsymbol{\mu} \times \boldsymbol{B} = -\frac{g}{2m}\boldsymbol{S} \times \boldsymbol{B}

$$

Rearranging with the **gyromagnetic ratio** $\gamma = g/(2m)$:

$$

\frac{d\boldsymbol{S}}{dt} = -\gamma \boldsymbol{S} \times \boldsymbol{B}

$$

or equivalently, defining $\boldsymbol{\mu} = \gamma \boldsymbol{S}$:

$$

\frac{d\boldsymbol{S}}{dt} = \gamma(\boldsymbol{B} \times \boldsymbol{S})

$$

**This is the Larmor precession equation!** It shows that the spin angular momentum vector precesses around the magnetic field direction.

### Larmor Precession

For a uniform magnetic field $\boldsymbol{B} = B\hat{z}$, the precession equation has an elegant geometric interpretation. The spin vector rotates around the $z$-axis with angular velocity:

$$

\boldsymbol{\Omega}_L = -\gamma\boldsymbol{B} = -\gamma B\hat{z}

$$

The **Larmor frequency** (in SI units) is:

$$

\omega_L = \gamma B = \frac{gB}{2m}

$$

In Gaussian units (often used in quantum mechanics texts):

$$

\omega_L = \frac{gB}{2mc}

$$

where $c$ is the speed of light. For an electron ($g = 2$, $m = m_e$):

$$

\omega_L = \frac{eB}{m_e c} \quad (\text{Gaussian units})

$$

The spin undergoes **uniform precession**: the magnitude $|\boldsymbol{S}|$ remains constant, but the vector traces out a cone around the $z$-axis with the tip at a fixed polar angle. If the spin is initially aligned at angle $\theta$ to the field, it maintains that angle while rotating around the field direction.

:::{admonition} Derivation: Larmor Precession from Torque Equation
:class: dropdown note

Starting from the torque equation:

$$

\frac{d\boldsymbol{S}}{dt} = -\gamma \boldsymbol{S} \times \boldsymbol{B}

$$

with $\boldsymbol{B} = B\hat{z}$, write $\boldsymbol{S} = (S_x, S_y, S_z)$:

$$

\frac{dS_x}{dt} = -\gamma(S_y B) = -\gamma B S_y

$$

$$

\frac{dS_y}{dt} = -\gamma(-S_x B) = \gamma B S_x

$$

$$

\frac{dS_z}{dt} = 0

$$

The $z$-component is conserved. For the $x$-$y$ components, define the complex variable $S_+ = S_x + iS_y$:

$$

\frac{dS_+}{dt} = -i\gamma B S_+

$$

This has the solution:

$$

S_+(t) = S_+(0)e^{-i\gamma B t}

$$

In terms of amplitude and phase:

$$

S_x(t) = |S_\perp|\cos(\theta_0 - \gamma B t), \quad S_y(t) = |S_\perp|\sin(\theta_0 - \gamma B t)

$$

where $|S_\perp| = \sqrt{S_x^2 + S_y^2}$ is the perpendicular component and $\theta_0$ is the initial azimuthal angle. The spin precesses around the $z$-axis with angular velocity $\omega_L = \gamma B$.

:::

### Energy and Stability

The magnetic energy of a spin in a field is:

$$

U = -\boldsymbol{\mu} \cdot \boldsymbol{B} = \gamma\boldsymbol{S} \cdot \boldsymbol{B}

$$

For a spin aligned with the field ($\boldsymbol{S} \parallel \boldsymbol{B}$): $U = -\gamma SB$ (minimum, stable).

For a spin anti-aligned ($\boldsymbol{S} \antiparallel \boldsymbol{B}$): $U = +\gamma SB$ (maximum, unstable).

The **energy splitting** between these states is:

$$

\Delta E = 2\gamma SB = gB \cdot \frac{S}{m}

$$

In quantum mechanics, this energy difference becomes the transition frequency for spin flips, which is observable in **electron spin resonance (ESR)** and **nuclear magnetic resonance (NMR)** experiments.


### Classical Analog of Quantum Spin

This classical picture of spin dynamics provides intuition for quantum spin:

1. **Quantization**: In quantum mechanics, the direction $\boldsymbol{n}$ becomes a quantum variable; only discrete orientations are allowed (spin quantization).

2. **Precession**: A spin in a magnetic field precesses at the Larmor frequency $\omega_L = gB/(2m_e c)$ where $g$ is the g-factor.

3. **Two-level system**: For spin-1/2, only two states (spin-up and spin-down) are possible, forming a qubit in quantum information.

4. **Berry phase**: The quantized spin undergoing adiabatic precession acquires a Berry phase—a topological phase that appears nowhere in the classical treatment.

### Thomas Precession

A subtle relativistic effect occurs when a spinning object undergoes acceleration: the spin axis precesses at a rate that differs from the orbital precession, even without an external magnetic field. This is **Thomas precession**.

For a non-relativistic electron in an atom, Thomas precession accounts for the difference between the orbital and total angular momentum contributions, modifying the spin-orbit coupling strength by a factor of 1/2.

In the rest frame of the electron, a moving nucleus creates an effective magnetic field (due to the relativistic transformation of electric field to magnetic field). The factor of 1/2 in the spin-orbit coupling $-\boldsymbol{\hat{\sigma}} \cdot \boldsymbol{B}_\text{eff}$ is purely a Thomas precession effect.

### Summary

- **Decomposition**: Angular momentum splits into axial component $L_\parallel$ and perpendicular part $\boldsymbol{L}_\perp$.
- **Spin axis**: Unit vector $\boldsymbol{n}$ specifies spin direction; evolves on the unit sphere.
- **Equation of motion**: $\frac{d\boldsymbol{n}}{dt} = \boldsymbol{n} \times \boldsymbol{\Omega}(\boldsymbol{n})$ resembles Lorentz force.
- **Precession**: Spin axis circles around effective field direction at rate $\boldsymbol{\Omega}$.
- **Lorentz analogy**: Spin dynamics maps to charged particle in magnetic field via sphere geometry.
- **Potential energy**: $V(\boldsymbol{n})$ determines effective field: $\boldsymbol{\Omega} = (\nabla V \times \boldsymbol{n})/I_t$.
- **Thomas precession**: Relativistic effect where spin axis precesses even without external field.
- **Quantum connection**: Classical intuition guides understanding of quantum spin quantization and Berry phase.
