## Lecture Notes

The central strategy of non-degenerate perturbation theory is to surgically separate a quantum system's evolution into two parts: the rapid, familiar oscillations of a solved Hamiltonian $H_0$, and the slow, subtle rearrangement caused by a perturbation $V$. This separation exposes a fundamental principle: energy gaps act as barriers to state mixing, with strong hybridization happening near resonance and exponential suppression far from it. The result is a systematic, order-by-order algorithm for computing energy shifts and state distortions without solving the full problem.

### Problem Setup

We consider the Hamiltonian:

$$

H(\lambda) = H_0 + \lambda V

$$ (eq-5b-ham-general)

**Assumptions**:

1. The spectrum of $H_0$ is **discrete with no degeneracies**: each eigenvalue appears exactly once.
2. The coupling parameter $\lambda$ is **small enough** that power series in $\lambda$ converges.
3. We know the complete spectrum of $H_0$: $\{E_n^{(0)}, |\psi_n^{(0)}\rangle\}$ for $n = 1, 2, 3, \ldots$
4. The perturbation $V$ is **Hermitian** and labeled by the same index $n$ throughout ("adiabatic continuity").

**Goal**: Expand eigenvalues and eigenstates as power series:

$$

E_n(\lambda) = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \ldots

$$ (eq-5b-energy-series)

$$|\psi_n(\lambda)\rangle = |\psi_n^{(0)}\rangle + \lambda |\psi_n^{(1)}\rangle + \lambda^2 |\psi_n^{(2)}\rangle + \ldots
$$ (eq-5b-state-series)

### The Hellmann-Feynman Theorem

The **Hellmann-Feynman theorem** is a cornerstone of perturbation theory. It states:

$$

\boxed{\frac{dE_n}{d\lambda} = \langle \psi_n(\lambda) | \frac{\partial H}{\partial\lambda} | \psi_n(\lambda) \rangle}

$$ (eq-5b-hf-theorem)

**Physical meaning**: The energy shifts with the perturbation parameter proportionally to the expectation value of the perturbation operator. Evaluated at $\lambda = 0$ with $\frac{\partial H}{\partial \lambda} = V$:

$$

\left.\frac{dE_n}{d\lambda}\right|_{\lambda=0} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle

$$ (eq-5b-hf-first-order)

This result shows that to first order, the energy shift is simply the expectation value of $V$ in the unperturbed state—a direct consequence of the adiabatic continuity assumption.

:::{admonition} Derivation
:class: dropdown note

Consider the eigenvalue equation:

$$

H(\lambda) |\psi_n(\lambda)\rangle = E_n(\lambda) |\psi_n(\lambda)\rangle

$$ (eq-5b-eigenvalue-eq)

Differentiate both sides with respect to $\lambda$:

$$

\frac{\partial H}{\partial \lambda}|\psi_n\rangle + H\frac{\partial |\psi_n\rangle}{\partial \lambda} = \frac{dE_n}{d\lambda} |\psi_n\rangle + E_n\frac{\partial |\psi_n\rangle}{\partial \lambda}

$$

Take the inner product with $\langle \psi_n(\lambda)|$:

$$

\langle \psi_n|\frac{\partial H}{\partial\lambda}|\psi_n\rangle + \langle \psi_n|H\frac{\partial |\psi_n\rangle}{\partial \lambda} = \frac{dE_n}{d\lambda}\langle\psi_n|\psi_n\rangle + E_n\langle\psi_n|\frac{\partial |\psi_n\rangle}{\partial \lambda}

$$

Use $H^\dagger = H$ (Hermiticity) to rewrite $\langle \psi_n|H = E_n\langle\psi_n|$:

$$

\langle \psi_n|\frac{\partial H}{\partial\lambda}|\psi_n\rangle + E_n\langle\psi_n|\frac{\partial |\psi_n\rangle}{\partial \lambda} = \frac{dE_n}{d\lambda} + E_n\langle\psi_n|\frac{\partial |\psi_n\rangle}{\partial \lambda}

$$

The $E_n$ terms cancel:

$$

\langle \psi_n|\frac{\partial H}{\partial\lambda}|\psi_n\rangle = \frac{dE_n}{d\lambda}

$$ (eq-5b-hf-main)

:::

### First-Order Energy Correction

The first-order energy correction is obtained by expanding the eigenvalue equation to order $\lambda^1$ and taking the inner product with the unperturbed state:

$$

\boxed{E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle}

$$ (eq-5b-E1-result)

**Physical interpretation**: The first-order energy shift is simply the **diagonal matrix element** of the perturbation in the unperturbed basis. It is the expectation value of $V$ in the unperturbed state. States with no diagonal matrix element (like $\hat{\sigma}^x$ in the computational basis) experience zero first-order shift.

:::{admonition} Derivation
:class: dropdown note

Expand $H|\psi_n\rangle = E_n|\psi_n\rangle$ to first order in $\lambda$:

$$

(H_0 + \lambda V)(|\psi_n^{(0)}\rangle + \lambda |\psi_n^{(1)}\rangle + \ldots) = (E_n^{(0)} + \lambda E_n^{(1)} + \ldots)(|\psi_n^{(0)}\rangle + \lambda |\psi_n^{(1)}\rangle + \ldots)

$$

**Order $\lambda^0$**:

$$

H_0 |\psi_n^{(0)}\rangle = E_n^{(0)} |\psi_n^{(0)}\rangle \quad \text{(unperturbed eigenvalue eq.)}

$$

**Order $\lambda^1$**:

$$

V|\psi_n^{(0)}\rangle + H_0|\psi_n^{(1)}\rangle = E_n^{(0)} |\psi_n^{(1)}\rangle + E_n^{(1)} |\psi_n^{(0)}\rangle

$$

Take the inner product with $\langle \psi_n^{(0)}|$:

$$

\langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle + \langle \psi_n^{(0)} | H_0 | \psi_n^{(1)} \rangle = E_n^{(0)} \langle \psi_n^{(0)} | \psi_n^{(1)} \rangle + E_n^{(1)}

$$

Use $\langle \psi_n^{(0)} | H_0 = E_n^{(0)} \langle \psi_n^{(0)} |$ and the orthogonality condition $\langle \psi_n^{(0)} | \psi_n^{(1)} \rangle = 0$:

$$

\langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle = E_n^{(1)}

$$

:::

:::{admonition} Normalization choice
:class: note

We implicitly choose the normalization $\langle \psi_n^{(0)} | \psi_n(\lambda) \rangle = 1$ (independent of $\lambda$). Expanding to first order gives $\langle \psi_n^{(0)} | \psi_n^{(1)} \rangle = 0$, so corrections are orthogonal to the unperturbed state.

:::

### First-Order State Correction

While the first-order energy is diagonal, the state mixes with other eigenstates. The first-order state correction is:

$$

|\psi_n^{(1)}\rangle = \sum_{m \neq n} \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} |\psi_m^{(0)}\rangle

$$ (eq-5b-psi1-result)

**Physical interpretation**: The perturbed state is a superposition of unperturbed states. The admixture of state $m$ is:
- **Proportional to the coupling**: $\langle m | V | n \rangle$ (off-diagonal matrix element)
- **Inversely proportional to the energy gap**: $1/(E_n^{(0)} - E_m^{(0)})$ (strong near-resonance; exponentially suppressed far from resonance)
- **Sign-determined by gap**: If state $m$ is higher in energy, the coefficient is positive (state is pulled toward $m$); if lower, negative

This is the **state hybridization** principle: quantum states mix strongest when they are close in energy and coupled.

:::{admonition} Derivation
:class: dropdown note

Expand $|\psi_n^{(1)}\rangle$ in the eigenbasis of $H_0$:

$$|\psi_n^{(1)}\rangle = \sum_{m \neq n} c_m^{(1)} |\psi_m^{(0)}\rangle$$

(We exclude $m=n$ by the orthogonality condition $\langle \psi_n^{(0)} | \psi_n^{(1)} \rangle = 0$.)

Insert into the order-$\lambda^1$ equation:

$$

V|\psi_n^{(0)}\rangle + H_0 \sum_m c_m^{(1)} |\psi_m^{(0)}\rangle = E_n^{(0)} \sum_m c_m^{(1)} |\psi_m^{(0)}\rangle + E_n^{(1)} |\psi_n^{(0)}\rangle

$$

Rearrange:

$$

V|\psi_n^{(0)}\rangle = \sum_m c_m^{(1)} (E_m^{(0)} - E_n^{(0)}) |\psi_m^{(0)}\rangle + E_n^{(1)} |\psi_n^{(0)}\rangle

$$

Project onto $\langle\psi_m^{(0)}|$ for $m \neq n$:

$$

\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle = c_m^{(1)} (E_m^{(0)} - E_n^{(0)})

$$

Solving for $c_m^{(1)}$:

$$

c_m^{(1)} = \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_m^{(0)} - E_n^{(0)}} = \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{-(E_n^{(0)} - E_m^{(0)})}

$$

Therefore:

$$

|\psi_n^{(1)}\rangle = \sum_{m \neq n} \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} |\psi_m^{(0)}\rangle

$$

:::

### Second-Order Energy Correction

The second-order energy correction captures "virtual transitions": the system temporarily borrows strength from (or lends to) intermediate states before returning. The result is:

$$

\boxed{E_n^{(2)} = \sum_{m \neq n} \frac{|\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle|^2}{E_n^{(0)} - E_m^{(0)}}}

$$ (eq-5b-E2-result)

**Key features**:

1. The numerator is always **non-negative** (absolute value squared of coupling strength).
2. The sign depends on the **denominator**: the energy difference $E_n^{(0)} - E_m^{(0)}$.
3. **For the ground state** ($n = 0$): All intermediate states $m$ satisfy $E_m^{(0)} > E_0^{(0)}$, so all denominators are **negative**. Therefore $E_0^{(2)} < 0$: **ground state energy is always lowered by perturbation**.
4. **For excited states**: Mix of positive contributions (from higher states) and negative (from lower states). Net effect depends on the spectrum.
5. **Convergence**: Series converges when $\lambda |\langle m | V | n \rangle| \ll \min_m |E_n^{(0)} - E_m^{(0)}|$, i.e., perturbation is small compared to energy gaps.

This order contributes to **level repulsion**: nearby levels push each other apart. The ground state borrows energy from excited states (lowering itself), which raises the excited state energies.

:::{admonition} Derivation
:class: dropdown note

Derive from the order-$\lambda^2$ eigenvalue equation. Project the order-$\lambda^1$ equation:

$$

V|\psi_n^{(0)}\rangle + H_0|\psi_n^{(1)}\rangle = E_n^{(0)} |\psi_n^{(1)}\rangle + E_n^{(1)} |\psi_n^{(0)}\rangle

$$

onto $\langle\psi_n^{(0)}|$ after applying $(H_0 - E_n^{(0)})$ to $|\psi_n^{(1)}\rangle$ and using the orthogonality $\langle \psi_n^{(0)} | \psi_n^{(1)} \rangle = 0$:

$$

\langle \psi_n^{(0)} | V | \psi_n^{(1)} \rangle = E_n^{(2)}

$$

Substitute the first-order state correction:

$$

E_n^{(2)} = \langle \psi_n^{(0)} | V | \psi_n^{(1)} \rangle = \sum_{m \neq n} \frac{\langle \psi_n^{(0)} | V | \psi_m^{(0)} \rangle \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}}

$$

Use the fact that $\langle \psi_n^{(0)} | V | \psi_m^{(0)} \rangle = \langle \psi_m^{(0)} | V^\dagger | \psi_n^{(0)} \rangle^* = \langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle^*$ (by Hermiticity):

$$

E_n^{(2)} = \sum_{m \neq n} \frac{|\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle|^2}{E_n^{(0)} - E_m^{(0)}}

$$

:::

#### Level Repulsion

The second-order term exhibits **level repulsion**: nearby energy levels are pushed apart by the perturbation.

- **Ground state**: $E_0^{(2)} < 0$ because all denominators are negative (all intermediate states are higher), so the ground state energy is lowered.
- **First excited state**: Receives positive contributions from ground state mixing (energy increases) and mixed contributions from higher states.
- **Net effect**: The spacing between levels increases:

$$

\Delta E(\lambda) = [E_1(\lambda) - E_0(\lambda)] > [E_1^{(0)} - E_0^{(0)}] = \Delta E(0)

$$ (eq-5b-level-repulsion)

:::{admonition} Level repulsion in words
:class: note

**Perturbation theory reveals that quantum levels repel each other.** This is a fundamental consequence of the coupling structure $|\langle m|V|n\rangle|^2$ (quadratic in coupling). The closer two levels are in energy, the stronger they couple, and the more they push apart. In non-degenerate systems, levels **never cross**—they execute an "avoided crossing."

:::

#### State Hybridization and Resonance

The perturbed eigenstate is a **hybrid** of multiple unperturbed states:

$$|\psi_n(\lambda)\rangle \approx |\psi_n^{(0)}\rangle + \sum_{m \neq n} \frac{\langle\psi_m^{(0)}|V|\psi_n^{(0)}\rangle}{E_n^{(0)} - E_m^{(0)}} |\psi_m^{(0)}\rangle + O(\lambda^2)
$$

**Mixing strength** depends on:

1. **Coupling matrix element**: $|\langle m|V|n\rangle|$ — larger coupling increases mixing.
2. **Energy gap**: $|E_n^{(0)} - E_m^{(0)}|$ — smaller gaps increase mixing exponentially (resonance effect).

**Examples**:
- **Covalent bonding in molecules**: Atomic orbitals of nearby energy hybridize to form molecular orbitals (bonding and antibonding).
- **Spin-orbit coupling in atoms**: Spin ($\uparrow/\downarrow$) and orbital angular momentum states mix when they are close in energy.
- **Stark effect in atoms**: Electric field ($V = eEz$) mixes states of opposite parity that have nearby energy.

#### Summary Table

| Quantity | Formula | Meaning |
|----------|---------|---------|
| **1st-order energy** | $E_n^{(1)} = \langle n^{(0)} \| V \| n^{(0)} \rangle$ | Expectation value of perturbation in unperturbed state |
| **2nd-order energy** | $E_n^{(2)} = \sum_{m\neq n} \frac{\|\langle m^{(0)}\|V\|n^{(0)}\rangle\|^2}{E_n^{(0)} - E_m^{(0)}}$ | Virtual transitions (borrowed from/lent to other levels) |
| **1st-order state** | $\|\psi_n^{(1)}\rangle = \sum_{m\neq n} \frac{\langle m^{(0)}\|V\|n^{(0)}\rangle}{E_n^{(0)} - E_m^{(0)}} \|m^{(0)}\rangle$ | Admixture of other states, weight $\propto \text{coupling}/\text{gap}$ |

**Convergence criterion**: Series converges when $\lambda \|V\| \ll \min_m(\|E_n^{(0)} - E_m^{(0)}\|)$, i.e., perturbation strength is much smaller than energy gaps.

:::{example} Example: The Qubit Revisited
:class: example

For the qubit model $H(\lambda) = \hat{\sigma}^z + \lambda\hat{\sigma}^x$:

**Unperturbed basis**:

$$|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \quad |1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}, \quad E_+^{(0)} = 1, \quad E_-^{(0)} = -1$$

**Perturbation matrix elements**:

$$

V = \hat{\sigma}^x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \Rightarrow\langle 0|\hat{\sigma}^x|0\rangle = 0, \quad \langle 1|\hat{\sigma}^x|1\rangle = 0, \quad \langle 0|\hat{\sigma}^x|1\rangle = 1

$$

**Energy of state $|0\rangle$ (labeled as $+$ index)**:

$$

E_+^{(1)} = \langle 0|\hat{\sigma}^x|0\rangle = 0

$$

$$

E_+^{(2)} = \frac{|\langle 1|\hat{\sigma}^x|0\rangle|^2}{E_+^{(0)} - E_-^{(0)}} = \frac{|1|^2}{1 - (-1)} = \frac{1}{2}

$$

Thus:

$$

E_+(\lambda) = 1 + 0 \cdot \lambda + \frac{1}{2}\lambda^2 + O(\lambda^3)

$$

Compare with exact: $E_+ = \sqrt{1+\lambda^2} = 1 + \frac{1}{2}\lambda^2 - \frac{1}{8}\lambda^4 + \ldots$ ✓ Matches through $O(\lambda^2)$!

**State of $|0\rangle$**:

$$|\psi_+^{(1)}\rangle = \frac{\langle 1|\hat{\sigma}^x|0\rangle}{E_+^{(0)} - E_-^{(0)}} |1\rangle = \frac{1}{2}|1\rangle$$

So:

$$|\psi_+(\lambda)\rangle = |0\rangle + \frac{\lambda}{2}|1\rangle + O(\lambda^2)$$

Again, matches the exact expansion from Section 5.1.1! ✓

:::
