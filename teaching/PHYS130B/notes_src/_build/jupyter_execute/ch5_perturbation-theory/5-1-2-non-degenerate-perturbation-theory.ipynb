## Lecture Notes

### Overview

Non-degenerate perturbation theory provides a systematic expansion of eigenvalues and eigenstates in powers of a small parameter $\lambda$, for the Hamiltonian $H(\lambda) = H_0 + \lambda V$. The **Hellmann-Feynman theorem** is the organizing principle: the exact rate at which an eigenvalue changes with $\lambda$ equals the expectation value of $V$ in the exact eigenstate.

### Hellmann-Feynman Theorem

:::{admonition} Hellmann-Feynman Theorem
:class: important

For a parameter-dependent Hamiltonian $H(\lambda)$ with eigenvalue $E_n(\lambda)$ and normalized eigenstate $|\psi_n(\lambda)\rangle$:

$$

\frac{\mathrm{d}E_n}{\mathrm{d}\lambda} = \langle \psi_n(\lambda) | \frac{\partial H}{\partial \lambda} | \psi_n(\lambda) \rangle

$$ (eq-5b-hellmann-feynman)

At $\lambda = 0$ with $\partial H/\partial\lambda = V$, this immediately gives the first-order energy:

$$

E_n^{(1)} = \frac{\mathrm{d}E_n}{\mathrm{d}\lambda}\bigg|_{\lambda=0} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle

$$

**Requires**: adiabatic continuity — eigenstates can be tracked continuously as $\lambda$ varies (no level crossings).
:::

:::{admonition} Derivation: Hellmann-Feynman Theorem
:class: dropdown information

Start from the eigenvalue equation $H(\lambda)|\psi_n(\lambda)\rangle = E_n(\lambda)|\psi_n(\lambda)\rangle$ and differentiate both sides with respect to $\lambda$:

$$

\frac{\partial H}{\partial\lambda}|\psi_n\rangle + H\frac{\partial|\psi_n\rangle}{\partial\lambda} = \frac{\mathrm{d}E_n}{\mathrm{d}\lambda}|\psi_n\rangle + E_n\frac{\partial|\psi_n\rangle}{\partial\lambda}

$$

Project onto $\langle\psi_n|$ from the left. Using $\langle\psi_n|H = E_n\langle\psi_n|$, the second term on the left becomes $E_n\langle\psi_n|\partial_\lambda\psi_n\rangle$, which cancels the second term on the right:

$$

\langle\psi_n|\frac{\partial H}{\partial\lambda}|\psi_n\rangle + E_n\langle\psi_n|\partial_\lambda\psi_n\rangle = \frac{\mathrm{d}E_n}{\mathrm{d}\lambda}\underbrace{\langle\psi_n|\psi_n\rangle}_{=1} + E_n\langle\psi_n|\partial_\lambda\psi_n\rangle

$$

The $E_n\langle\psi_n|\partial_\lambda\psi_n\rangle$ terms cancel, leaving

$$

\frac{\mathrm{d}E_n}{\mathrm{d}\lambda} = \langle\psi_n(\lambda)|\frac{\partial H}{\partial\lambda}|\psi_n(\lambda)\rangle

$$

**Key insight**: the state derivatives $|\partial_\lambda\psi_n\rangle$ drop out entirely — the energy derivative depends only on the *current* eigenstate and the *explicit* parameter dependence of $H$.
:::

### First-Order Energy Correction

:::{admonition} First-Order Energy
:class: important

$$

\boxed{E_n^{(1)} = \langle \psi_n^{(0)} | V | \psi_n^{(0)} \rangle}

$$ (eq-5b-E1-result)

The first-order shift is the **diagonal matrix element** of $V$ in the unperturbed basis — the expectation value of the perturbation in the unperturbed state.
:::

If $V$ has no diagonal elements (e.g., $\hat{\sigma}^x$ in the computational basis), the first-order correction vanishes and the leading effect is second-order.

:::{admonition} Derivation: First-Order Energy from the Eigenvalue Equation
:class: dropdown information

Write the perturbative expansions $E_n = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \cdots$ and $|\psi_n\rangle = |\psi_n^{(0)}\rangle + \lambda|\psi_n^{(1)}\rangle + \cdots$. Substituting into $H|\psi_n\rangle = E_n|\psi_n\rangle$:

$$

(H_0 + \lambda V)(|\psi_n^{(0)}\rangle + \lambda|\psi_n^{(1)}\rangle + \cdots) = (E_n^{(0)} + \lambda E_n^{(1)} + \cdots)(|\psi_n^{(0)}\rangle + \lambda|\psi_n^{(1)}\rangle + \cdots)

$$

**Order $\lambda^0$**: $H_0|\psi_n^{(0)}\rangle = E_n^{(0)}|\psi_n^{(0)}\rangle$ (the unperturbed problem).

**Order $\lambda^1$**: $V|\psi_n^{(0)}\rangle + H_0|\psi_n^{(1)}\rangle = E_n^{(0)}|\psi_n^{(1)}\rangle + E_n^{(1)}|\psi_n^{(0)}\rangle$.

Project the order-$\lambda^1$ equation onto $\langle\psi_n^{(0)}|$:

$$

\langle\psi_n^{(0)}|V|\psi_n^{(0)}\rangle + \underbrace{\langle\psi_n^{(0)}|H_0|\psi_n^{(1)}\rangle}_{= E_n^{(0)}\langle\psi_n^{(0)}|\psi_n^{(1)}\rangle} = E_n^{(0)}\langle\psi_n^{(0)}|\psi_n^{(1)}\rangle + E_n^{(1)}

$$

The $E_n^{(0)}\langle\psi_n^{(0)}|\psi_n^{(1)}\rangle$ terms cancel (regardless of whether $\langle\psi_n^{(0)}|\psi_n^{(1)}\rangle$ vanishes), giving $E_n^{(1)} = \langle\psi_n^{(0)}|V|\psi_n^{(0)}\rangle$.

**Normalization convention**: we choose $\langle\psi_n^{(0)}|\psi_n^{(1)}\rangle = 0$ (the first-order correction is orthogonal to the unperturbed state). This fixes the phase ambiguity and ensures $|\psi_n\rangle$ remains normalized through first order: $\langle\psi_n|\psi_n\rangle = 1 + \lambda(\langle\psi_n^{(0)}|\psi_n^{(1)}\rangle + \text{c.c.}) + O(\lambda^2) = 1 + O(\lambda^2)$.
:::

### First-Order State Correction

:::{admonition} First-Order State
:class: important

$$

\boxed{|\psi_n^{(1)}\rangle = \sum_{m \neq n} \frac{\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} |\psi_m^{(0)}\rangle}

$$ (eq-5b-psi1-result)

The admixture of state $m$ is proportional to the **coupling** $\langle m|V|n\rangle$ and inversely proportional to the **energy gap** $E_n^{(0)} - E_m^{(0)}$. States mix strongest when they are close in energy and strongly coupled.
:::

:::{admonition} Derivation: First-Order State Correction
:class: dropdown information

From the order-$\lambda^1$ equation:

$$

V|\psi_n^{(0)}\rangle + H_0|\psi_n^{(1)}\rangle = E_n^{(0)}|\psi_n^{(1)}\rangle + E_n^{(1)}|\psi_n^{(0)}\rangle

$$

Rearrange: $(E_n^{(0)} - H_0)|\psi_n^{(1)}\rangle = V|\psi_n^{(0)}\rangle - E_n^{(1)}|\psi_n^{(0)}\rangle$.

Expand $|\psi_n^{(1)}\rangle$ in the unperturbed basis: $|\psi_n^{(1)}\rangle = \sum_{m\neq n} c_m^{(1)}|\psi_m^{(0)}\rangle$ (the $m=n$ component vanishes by our normalization convention). Projecting onto $\langle\psi_m^{(0)}|$ with $m\neq n$:

$$

(E_n^{(0)} - E_m^{(0)})\,c_m^{(1)} = \langle\psi_m^{(0)}|V|\psi_n^{(0)}\rangle - E_n^{(1)}\underbrace{\langle\psi_m^{(0)}|\psi_n^{(0)}\rangle}_{=0}

$$

Solving:

$$

c_m^{(1)} = \frac{\langle\psi_m^{(0)}|V|\psi_n^{(0)}\rangle}{E_n^{(0)} - E_m^{(0)}}

$$

This requires $E_n^{(0)} \neq E_m^{(0)}$ for all $m\neq n$ — the **non-degeneracy condition**. If any denominator vanishes, we must switch to degenerate perturbation theory (\S5.1.3).

**Tip on the energy denominator**: the energy of the state being *corrected* ($E_n$) always appears in the numerator position minus the energy of the *admixed* state ($E_m$).
:::

### Second-Order Energy Correction

:::{admonition} Second-Order Energy
:class: important

$$

\boxed{E_n^{(2)} = \sum_{m \neq n} \frac{|\langle \psi_m^{(0)} | V | \psi_n^{(0)} \rangle|^2}{E_n^{(0)} - E_m^{(0)}}}

$$ (eq-5b-E2-result)

- **Ground state**: all denominators are negative, so $E_0^{(2)} \leq 0$ — perturbations always lower the ground state energy (consistent with the variational principle).
- **Level repulsion**: nearby levels push each other apart through virtual transitions.
- **Convergence**: requires $\lambda|\langle m|V|n\rangle| \ll |E_n^{(0)} - E_m^{(0)}|$ for all $m$.
:::

:::{admonition} Derivation: Second-Order Energy Correction
:class: dropdown information

**Method 1 (direct):** From the order-$\lambda^2$ equation, project onto $\langle\psi_n^{(0)}|$:

$$

E_n^{(2)} = \langle\psi_n^{(0)}|V|\psi_n^{(1)}\rangle

$$

(The $H_0|\psi_n^{(2)}\rangle$ and $E_n^{(0)}|\psi_n^{(2)}\rangle$ terms cancel as before.) Substituting the first-order state:

$$

E_n^{(2)} = \sum_{m\neq n} \frac{\langle\psi_n^{(0)}|V|\psi_m^{(0)}\rangle\langle\psi_m^{(0)}|V|\psi_n^{(0)}\rangle}{E_n^{(0)} - E_m^{(0)}} = \sum_{m\neq n} \frac{|V_{mn}|^2}{E_n^{(0)} - E_m^{(0)}}

$$

**Method 2 (Hellmann-Feynman):** Differentiate $\mathrm{d}E_n/\mathrm{d}\lambda = \langle\psi_n|V|\psi_n\rangle$ again:

$$

\frac{\mathrm{d}^2 E_n}{\mathrm{d}\lambda^2} = \langle\partial_\lambda\psi_n|V|\psi_n\rangle + \langle\psi_n|V|\partial_\lambda\psi_n\rangle

$$

Insert the resolution of identity $\sum_m |\psi_m\rangle\langle\psi_m| = \hat{I}$, separate $m=n$ from $m\neq n$:

$$

\frac{\mathrm{d}^2 E_n}{\mathrm{d}\lambda^2}\bigg|_{\lambda=0} = \sum_{m\neq n}\left(\frac{V_{nm}}{E_n - E_m}V_{mn} + V_{nm}\frac{V_{mn}}{E_n - E_m}\right) + V_{nn}\underbrace{\partial_\lambda\langle\psi_n|\psi_n\rangle}_{=0}

$$

The $m=n$ term vanishes because $\partial_\lambda\langle n|n\rangle = 0$ (normalization is preserved). The result is $\mathrm{d}^2 E_n/\mathrm{d}\lambda^2 = 2\sum_{m\neq n}|V_{mn}|^2/(E_n - E_m)$, giving $E_n^{(2)} = \tfrac{1}{2}\mathrm{d}^2 E_n/\mathrm{d}\lambda^2 = \sum_{m\neq n}|V_{mn}|^2/(E_n - E_m)$.

**Ground state bound**: for the ground state ($n=0$), every denominator $E_0 - E_m < 0$, so $E_0^{(2)} \leq 0$. This is consistent with the variational principle: any trial state gives an energy $\geq$ the true ground state energy.
:::

### Physical Intuitions

The perturbation series has a clean physical interpretation at each order:

- **First-order energy** = expectation value of the perturbation (average energy shift in the unperturbed state)
- **First-order state** = hybridization with other levels, weighted by coupling/gap
- **Second-order energy** = virtual transitions to intermediate states and back; always lowers the ground state
- **Breakdown**: when $E_n^{(0)} - E_m^{(0)} \to 0$ (degeneracy), denominators diverge → signal to switch to degenerate perturbation theory (\S5.1.3)

:::{admonition} Example: Verification Against the Qubit Model
:class: dropdown example

**Problem.** Verify perturbation theory on the qubit $H(\lambda) = \hat{\sigma}^z + \lambda\hat{\sigma}^x$ from \S5.1.1.

**Solution.** Unperturbed: $E_+^{(0)} = +1$ ($|0\rangle$), $E_-^{(0)} = -1$ ($|1\rangle$). Since $\langle 0|\hat{\sigma}^x|0\rangle = 0$, the first-order energy vanishes: $E_+^{(1)} = 0$. The second-order correction:

$$

E_+^{(2)} = \frac{|\langle 1|\hat{\sigma}^x|0\rangle|^2}{E_+^{(0)} - E_-^{(0)}} = \frac{1}{2}

$$

So $E_+(\lambda) = 1 + \tfrac{1}{2}\lambda^2 + O(\lambda^3)$, matching the exact result $\sqrt{1+\lambda^2} = 1 + \tfrac{1}{2}\lambda^2 - \tfrac{1}{8}\lambda^4 + \cdots$. $\checkmark$

The first-order state: $|\psi_+^{(1)}\rangle = \tfrac{1}{2}|1\rangle$, so $|\psi_+(\lambda)\rangle \approx |0\rangle + \tfrac{\lambda}{2}|1\rangle$, again matching the exact eigenvector expansion. $\checkmark$
:::

:::{admonition} Derivation: Second-Order State Correction
:class: dropdown information

Collecting the order-$\lambda^2$ terms from the eigenvalue equation and projecting onto $\langle\psi_m^{(0)}|$ with $m\neq n$:

$$

c_m^{(2)} = \sum_{l\neq n} \frac{V_{ml}\,V_{ln}}{(E_n - E_m)(E_n - E_l)} - \frac{V_{mn}\,V_{nn}}{(E_n - E_m)^2}

$$

The full second-order state also acquires a correction along $|\psi_n^{(0)}\rangle$ itself (from re-normalization):

$$

|\psi_n^{(2)}\rangle = \sum_{m\neq n} c_m^{(2)}|\psi_m^{(0)}\rangle - \frac{1}{2}\sum_{m\neq n}\frac{|V_{mn}|^2}{(E_n - E_m)^2}|\psi_n^{(0)}\rangle

$$

The last term ensures normalization through second order: $\langle\psi_n|\psi_n\rangle = 1 + O(\lambda^3)$. It arises from differentiating $\langle\psi_n^{(0)}|\psi_n^{(0)}\rangle = 1$ and requiring $\langle\psi_n^{(0)}|\psi_n^{(2)}\rangle + \langle\psi_n^{(1)}|\psi_n^{(1)}\rangle = 0$, which gives

$$

\langle\psi_n^{(0)}|\psi_n^{(2)}\rangle = -\frac{1}{2}\langle\psi_n^{(1)}|\psi_n^{(1)}\rangle = -\frac{1}{2}\sum_{m\neq n}\frac{|V_{mn}|^2}{(E_n-E_m)^2}

$$

:::

:::{admonition} Discussion
:class: dropdown tip

The second-order formula $E_n^{(2)} = \sum_m |V_{mn}|^2/(E_n - E_m)$ diverges when two levels become degenerate ($E_n = E_m$). Is this a failure of quantum mechanics, or just a signal that you need a different computational method? What is the correct procedure when the energy gap $\Delta E$ approaches zero while the coupling $V_{mn}$ stays finite?
:::

### Summary

- The **Hellmann-Feynman theorem** $\mathrm{d}E_n/\mathrm{d}\lambda = \langle\psi_n|V|\psi_n\rangle$ is the organizing principle: energy derivatives are expectation values.
- **First-order energy**: $E_n^{(1)} = \langle n^{(0)}|V|n^{(0)}\rangle$ — expectation value of perturbation.
- **First-order state**: admixture of other levels, weighted by coupling over energy gap.
- **Second-order energy**: virtual transitions; always lowers the ground state; series diverges at degeneracy.
