## Lecture Notes

Fermi's golden rule distills the essence of quantum transitions: when a system couples weakly to a dense spectrum of final states, the transition probability becomes a timeless rate, linear in time and proportional to the squared coupling strength times the density of available targets. This rule underlies atomic spectroscopy, radioactive decay, and particle scattering—it is the bridge between perturbation theory's formal apparatus and the measurable transition rates that define real quantum phenomena.

### Transition Probability: First-Order Calculation

#### Definition

Consider a system initially in state $|i\rangle$ (eigenstate of $H_0$ with energy $E_i^{(0)}$). The **transition probability** to reach state $|f\rangle$ (energy $E_f^{(0)}$) at time $t$ is:

$$

P_{i\to f}(t) = |\langle f | U_S(t) | i \rangle|^2

$$ (eq-5b-prob-def)

Using the factorization $U_S(t) = e^{-iH_0 t/\hbar} U_I(t)$, the overall phase cancels and we get:

$$

P_{i\to f}(t) = |\langle f | U_I(t) | i \rangle|^2

$$ (eq-5b-prob-IP)

To **first order** in the coupling $\lambda$ (valid for $|V| t / \hbar \ll 1$), using the first-order term of the Dyson series:

$$

\boxed{P_{i\to f}^{(1)}(t) = \frac{1}{\hbar^2} \left| \int_0^t dt_1 e^{i\omega_{fi} t_1} V_{fi}(t_1) \right|^2}

$$ (eq-5b-P1st-general)

where $V_{fi}(t) = \langle f | V(t) | i \rangle$ is the matrix element, and:

$$

\omega_{fi} = \frac{E_f^{(0)} - E_i^{(0)}}{\hbar}

$$ (eq-5b-omega-fi)

is the **transition frequency** (resonance is at $\omega_{fi} = 0$, i.e., $E_f = E_i$).

### Sinc Function and Resonance

#### Time-Independent Perturbation: Constant Coupling

For a **constant (time-independent) perturbation** $V(t) = V_0$ applied for a duration $t$:

$$

\boxed{P_{i\to f}^{(1)}(t) = |V_{fi}|^2 \frac{\sin^2(\omega_{fi} t/2)}{\hbar^2 (\omega_{fi}/2)^2} = |V_{fi}|^2 t^2 \operatorname{sinc}^2\left(\frac{\omega_{fi} t}{2}\right)}

$$ (eq-5b-P-sinc-formula)

where $\operatorname{sinc}(x) = \sin(x)/x$ is the sinc function.

**Physical interpretation**:
- **On resonance** ($\omega_{fi} \to 0$, i.e., $E_f \to E_i$): $P \approx |V_{fi}|^2 t^2/\hbar^2$, which grows **quadratically in time**. This implies a **constant transition rate** $\Gamma = dP/dt \approx 2|V_{fi}|^2 t/\hbar^2$ (linear in $t$).
- **Off-resonance** ($|\omega_{fi}| \gg 1/t$): The sinc function oscillates rapidly and averages to a small value $\propto 1/\omega_{fi}^2$, **suppressing the transition**.

The sinc function has **zero crossings** at $\omega_{fi} = 2\pi n / t$, meaning that even off-resonant energies can give destructive interference. The **width of the main lobe** is $\Delta \omega \sim 2\pi / t$ (resolution improves with longer observation time).

:::{admonition} Derivation
:class: dropdown note

For constant $V_0$:

$$

P_{i\to f}^{(1)}(t) = \frac{|V_{fi}|^2}{\hbar^2} \left| \int_0^t dt_1 e^{i\omega_{fi}t_1} \right|^2

$$

Evaluate the integral:

$$

\int_0^t dt_1 e^{i\omega_{fi}t_1} = \frac{e^{i\omega_{fi}t} - 1}{i\omega_{fi}}

$$

For $\omega_{fi} \neq 0$:

$$

\left|\int_0^t dt_1 e^{i\omega_{fi}t_1}\right|^2 = \frac{(e^{i\omega_{fi}t} - 1)(e^{-i\omega_{fi}t} - 1)}{\omega_{fi}^2} = \frac{2 - 2\cos(\omega_{fi}t)}{\omega_{fi}^2} = \frac{4\sin^2(\omega_{fi}t/2)}{\omega_{fi}^2}

$$

Therefore:

$$

P_{i\to f}^{(1)}(t) = |V_{fi}|^2 \frac{\sin^2(\omega_{fi}t/2)}{\hbar^2(\omega_{fi}/2)^2}

$$

:::

**Physical picture**: Imagine shining a laser at a specific frequency on an atom. If the laser frequency matches the atomic transition ($\omega_{fi} = 0$), absorption is strong and builds up over time. If the laser is detuned (off-resonance), the oscillating coupling causes the atom to absorb and re-emit, with the oscillations mostly canceling out. Only transitions within a bandwidth $\Delta\omega \sim 2\pi/T$ (where $T$ is the observation time) give net probability.

### Fermi's Golden Rule

#### From Sinc to Delta Function: Long-Time Limit

In real systems, the final state is usually part of a **continuum** or **dense spectrum**: scattering into all possible momenta, decay into photons with all energies, absorption across a band of states. We do not ask "what is the probability to reach a specific state $|f\rangle$?" but rather "what is the total transition rate into any state near energy $E_i$?"

The sinc function becomes increasingly sharply peaked at $\omega_{fi} = 0$ as $t \to \infty$. In the limit:

$$

\lim_{t \to \infty} \frac{t \sin^2(\omega_{fi} t/2)}{(\omega_{fi}/2)^2} = \pi t \delta(\omega_{fi})

$$

This delta function enforces **energy conservation**: only final states with $E_f = E_i$ contribute.

#### Derivation and Statement

Consider the total transition probability into all final states within an energy interval:

$$

P_{i \to \text{continuum}}(t) = \sum_{f} P_{i\to f}(t) \approx \int dE_f \, \rho(E_f) P_{i\to f}^{(1)}(t)

$$

where $\rho(E_f)$ is the **density of states** (number of states per unit energy).

In the **long-time limit** $t \to \infty$:

$$

P_{i\to\text{continuum}}(t) \approx |V_{fi}|^2 \rho(E_i) \frac{2\pi t}{\hbar}

$$

The transition **rate** (probability per unit time) is constant:

$$

\boxed{\Gamma = \frac{dP}{dt} = \frac{2\pi}{\hbar} |V_{fi}|^2 \rho(E_i)}

$$ (eq-5b-fgr-standard)

**Fermi's Golden Rule** (the fundamental result):

$$

\boxed{\Gamma_{i\to f} = \frac{2\pi}{\hbar} |\langle f | V | i \rangle|^2 \delta(E_f - E_i)}

$$ (eq-5b-fgr-detailed)

:::{admonition} Derivation
:class: dropdown note

Start from the sinc-function formula:

$$

P_{i\to f}^{(1)}(t) = |V_{fi}|^2 \frac{\sin^2(\omega_{fi}t/2)}{\hbar^2(\omega_{fi}/2)^2}

$$

**Sum over final states** in a continuum with density $\rho(E_f)$:

$$

P_{i\to\text{continuum}}(t) = \int dE_f \, \rho(E_f) |V_{fi}(E_f)|^2 \frac{\sin^2(\omega_{fi}t/2)}{\hbar^2(\omega_{fi}/2)^2}

$$

Change variables to $\omega_{fi} = (E_f - E_i)/\hbar$, so $dE_f = \hbar \, d\omega_{fi}$:

$$

P_{i\to\text{continuum}}(t) = \int d\omega_{fi} \, \rho(E_i + \hbar\omega_{fi}) |V_{fi}(\omega_{fi})|^2 \frac{t \sin^2(\omega_{fi}t/2)}{\hbar(\omega_{fi}/2)^2}

$$

**In the long-time limit** $t \to \infty$, the product $t \sin^2(\omega_{fi}t/2) / (\omega_{fi}/2)^2$ becomes sharply peaked around $\omega_{fi} = 0$ with width $\sim 2\pi/t \to 0$. Using the delta-function identity:

$$

\lim_{t\to\infty} \frac{t\sin^2(\omega_{fi}t/2)}{(\omega_{fi}/2)^2} = \pi t \delta(\omega_{fi})

$$

we get:

$$

P_{i\to\text{continuum}}(t) = \int d\omega_{fi} \, \rho(E_i) |V_{fi}|^2 \frac{\pi t}{\hbar} \delta(\omega_{fi})

$$

The delta function forces $\omega_{fi} = 0$:

$$

P_{i\to\text{continuum}}(t) = \rho(E_i) |V_{fi}|^2 \frac{\pi t}{\hbar}

$$

**Transition rate**:

$$

\Gamma = \frac{dP}{dt} = \rho(E_i) |V_{fi}|^2 \frac{\pi}{\hbar}

$$

(The factor differs slightly depending on normalization conventions; standard is $2\pi/\hbar$.)

:::

#### Physical Interpretation

**Fermi's Golden Rule states that the transition rate is proportional to:**

1. **The square of the matrix element** $|\langle f | V | i \rangle|^2$: Stronger coupling → faster transitions.
2. **The density of final states** $\rho(E_i)$: More available target states → more transition channels → faster overall decay or absorption.
3. **Energy conservation** $\delta(E_f - E_i)$: Only states with the same energy as the initial state contribute (elastic scattering or energy-conserving processes).

In loose terms: **Rate $\propto$ (coupling strength)$^2$ $\times$ (number of available targets).**

### Applications

#### Spontaneous Emission (Radiative Decay)

An excited atomic state $|i\rangle$ decays to lower states by emitting photons:

$$

\Gamma_{\text{decay}} = \frac{2\pi}{\hbar} |\langle f | \boldsymbol{d} \cdot \boldsymbol{E} | i \rangle|^2 \rho_\gamma(E_f)

$$

where $\rho_\gamma$ is the **photon density of states** (number of radiation modes). The **lifetime** is $\tau = 1/\Gamma$.

#### Scattering Cross Section

In particle scattering, the final state is a plane wave (continuum of momenta). Fermi's rule relates the **scattering rate** to the **Born approximation** (first-order scattering amplitude). The **differential cross section** is:

$$

\frac{d\sigma}{d\Omega} \propto |\langle\boldsymbol{k}_f | V | \boldsymbol{k}_i \rangle|^2

$$

#### Photoabsorption

A photon of energy $\hbar\omega$ excites an electron across a band gap. The **absorption rate** depends on:
- The density of initial states (valence band)
- The density of final states (conduction band)
- The photon energy (resonance if $\hbar\omega$ matches the band gap)
- The dipole matrix element (how strongly light couples)

Fermi's rule quantifies all these factors.

### Validity Conditions and Assumptions

Fermi's golden rule applies when:

1. **First-order approximation**: $|V| t / \hbar \ll 1$ (weak perturbation, short times) or $\rho |V|^2 \ll 1$ (coupling much weaker than density-of-states scale).
2. **Long-time limit**: $t \gg 1/|\omega_{fi}|$ (evolution time much longer than inverse transition frequency). This allows the sinc function to sharpen into a delta function.
3. **Continuous or dense final-state spectrum**: Final states must form a continuum or be densely packed. If final states are discrete and widely spaced, the delta function is not appropriate.
4. **No initial-final overlap**: $\langle i | f \rangle = 0$ or negligible (initial and final states are orthogonal).

### Selection Rules and Forbidden Transitions

#### Symmetry-Based Constraints

The matrix element $\langle f | V | i \rangle$ **vanishes identically** if $V$ and the basis states have incompatible symmetries. These are **selection rules**.

**Example: Electric dipole radiation** ($V = -\boldsymbol{d} \cdot \boldsymbol{E}$, dipole moment coupled to field)

Selection rules:
- $\Delta \ell = \pm 1$ (orbital angular momentum must change by 1)
- $\Delta m_\ell = 0, \pm 1$ (magnetic quantum number; depends on light polarization)
- $\Delta S = 0$ (spin is not affected in lowest order; spin-flip requires spin-orbit coupling)

If a transition **violates these rules**, the dipole matrix element is zero, and the transition is **forbidden to leading order** (first-order dipole transitions). It may proceed at a much lower rate via higher-order processes (electric quadrupole, magnetic dipole, etc.).

#### Calculating Transition Rates

Practical steps:
1. Identify the perturbation $V$ (e.g., $\boldsymbol{d} \cdot \boldsymbol{E}$ for photon coupling).
2. Write the initial $|i\rangle$ and final $|f\rangle$ states with all quantum numbers.
3. Evaluate $\langle f | V | i \rangle$. Check selection rules.
4. Determine $\rho(E_f)$ at the relevant energy.
5. Apply: $\Gamma = (2\pi/\hbar) |\langle f|V|i\rangle|^2 \rho(E_i)$.

### Summary

- **First-order transition probability**: Oscillates as $\sin^2(\omega_{fi}t/2) / \omega_{fi}^2$. Enhanced on resonance ($\omega_{fi} \approx 0$), suppressed off-resonance.
- **Sinc function**: Encodes destructive interference for off-resonant coupling. Main lobe width $\sim 2\pi/t$.
- **Fermi's Golden Rule**: $\Gamma = (2\pi/\hbar) |V_{fi}|^2 \rho(E_i) \delta(E_f - E_i)$ (with proper energy normalization).
- **Key insight**: Rate is constant in the long-time limit for a continuum spectrum; depends on coupling squared and density of states.
- **Validity**: Weak perturbation, long time, continuous spectrum, first-order approximation valid.
- **Selection rules**: Symmetry constraints determine which transitions are allowed. Violation = forbidden transition (suppressed rate).
- **Applications**: Radiative decay (lifetime), scattering (cross section), absorption (transition rates in materials).
