## Lecture Notes

Perturbation theory is not an abstract tool but the language of quantum control and quantum observation: harmonic driving creates Rabi oscillations and quantum gates; adiabatic evolution harvests geometric phases and enables population transfer; Green's functions and scattering amplitudes encode the structure of bound and continuum states; Feynman diagrams transform infinite perturbative series into intuitive pictures of virtual processes. These applications reveal perturbation theory as the practical foundation of quantum engineering and measurement.

### Application 1: Harmonic (Oscillating) Perturbations

#### Physical Setup

Many real perturbations are **time-harmonic** (oscillating), such as:
- Electromagnetic waves: $\boldsymbol{E}(t) = \boldsymbol{E}_0 \cos(\omega t + \phi)$ (light, radio waves)
- Driving fields in quantum control: AC gate voltages on qubits
- Interaction with phonons or photons in condensed matter

A harmonic perturbation can be written as:

$$


V(t) = V_0 \cos(\omega t) = \frac{V_0}{2} (\mathrm{e}^{-\mathrm{i}\omega t} + \mathrm{e}^{\mathrm{i}\omega t})

$$ (eq-5c-harmonic-V)

or more generally, in the rotating-wave representation:

$$


V(t) = (V \mathrm{e}^{-\mathrm{i}\omega t} + V^\dagger \mathrm{e}^{\mathrm{i}\omega t})

$$ (eq-5c-harmonic-V-gen)

#### Transition Amplitude and Resonance

Using first-order perturbation theory from the Dyson series:

$$


\mathcal{A}_{i\to f}^{(1)}(t) = -\frac{\mathrm{i}}{\hbar} \int_0^t dt_1 \mathrm{e}^{\mathrm{i}\omega_{fi}t_1} \langle f | V(t_1) | i \rangle

$$


For harmonic $V(t) = \frac{V_0}{2}(\mathrm{e}^{-\mathrm{i}\omega t} + \mathrm{e}^{\mathrm{i}\omega t})$, this integral contains two terms with frequencies $\omega_{fi} \pm \omega$.

**Resonance occurs at**:
1. **Absorption** ($E_f > E_i$): $\omega = \omega_{fi} = (E_f - E_i)/\hbar$ ‚ÄĒ the photon energy matches the energy gap; this term accumulates.
2. **Stimulated emission** ($E_f < E_i$): $\omega = -\omega_{fi}$ ‚ÄĒ energy is released back to the field; this term accumulates when driving at the emission frequency.

Away from resonance, the two terms oscillate at very different frequencies and destructively interfere, suppressing transitions. **This is the essence of the rotating wave approximation**.

#### Rabi Oscillations

In the **rotating wave approximation** (neglecting rapidly oscillating off-resonant terms) and on **exact resonance** ($\omega = \omega_{fi}$), the transition probability is:

$$


\boxed{P_{i\to f}(t) = \frac{|V_{fi}|^2}{\hbar^2} \sin^2\left(\frac{\Omega_R t}{2}\right)}

$$ (eq-5c-P-rabi)

:::{admonition} Rabi Frequency
:class: important

The **Rabi frequency** is:

$$


\boxed{\Omega_R = \frac{|V_{fi}|}{\hbar}}

$$ (eq-5c-rabi-freq)

This represents the rate at which the system oscillates between the two states. Doubling the coupling strength (matrix element) doubles the Rabi frequency.

:::

:::{admonition} Physical Interpretation: Coherent Oscillations
:class: note

The probability **oscillates sinusoidally** between 0 and 1 as the system coherently exchanges energy with the driving field. This is **Rabi oscillation**. Starting in state $|i\rangle$, the system oscillates between $|i\rangle$ and $|f\rangle$, reaching maximum probability (population inversion) at times $t = \pi/(2\Omega_R)$.

**Applications**:
- **NMR resonance**: Radiofrequency (RF) pulses drive spin transitions; Rabi frequency determines pulse duration.
- **Cavity QED**: Atoms driven by cavity photons exhibit Rabi oscillations.
- **Quantum computing**: Rabi pulses form the basis of quantum logic gates. A "ŌÄ-pulse" (rotating $\pi$ radians) swaps populations.
- **Two-level approximation**: Any system can be treated as a two-level system near a resonance, with Rabi oscillations describing the dynamics.

:::{admonition} Derivation
:class: dropdown note

Start with:

$$


\mathcal{A}_{i\to f}^{(1)}(t) = -\frac{\mathrm{i}}{\hbar} \int_0^t dt_1 \mathrm{e}^{\mathrm{i}\omega_{fi}t_1} \langle f | V(t_1) | i \rangle

$$


Insert $V(t) = \frac{V_0}{2}(\mathrm{e}^{-\mathrm{i}\omega t} + \mathrm{e}^{\mathrm{i}\omega t})$ where $V_0 = V_{fi}$ is the matrix element:

$$


\mathcal{A}_{i\to f}^{(1)}(t) = -\frac{\mathrm{i} V_0}{2\hbar} \left[ \int_0^t dt_1 \mathrm{e}^{\mathrm{i}(\omega_{fi} - \omega)t_1} + \int_0^t dt_1 \mathrm{e}^{\mathrm{i}(\omega_{fi} + \omega)t_1} \right]

$$


Evaluate the integrals:

$$


\int_0^t dt_1 \mathrm{e}^{\mathrm{i}\Delta\omega t_1} = \frac{\mathrm{e}^{\mathrm{i}\Delta\omega t} - 1}{i\Delta\omega}

$$


giving:

$$


\mathcal{A}_{i\to f}^{(1)}(t) = -\frac{\mathrm{i} V_0}{2\hbar} \left[ \frac{\mathrm{e}^{\mathrm{i}(\omega_{fi}-\omega)t} - 1}{i(\omega_{fi} - \omega)} + \frac{\mathrm{e}^{\mathrm{i}(\omega_{fi}+\omega)t} - 1}{i(\omega_{fi} + \omega)} \right]

$$


**Rotating wave approximation**: If $\omega \approx \omega_{fi}$, the first term is resonant (slow or zero oscillation), while the second term oscillates rapidly at $\approx 2\omega_{fi}$. Neglecting the rapid term:

$$


\mathcal{A}_{i\to f}(t) \approx -\frac{\mathrm{i} V_0}{2\hbar} \cdot \frac{\mathrm{e}^{\mathrm{i}(\omega_{fi}-\omega)t} - 1}{i(\omega_{fi} - \omega)} = \frac{V_0}{2\hbar} \cdot \frac{\mathrm{e}^{\mathrm{i}(\omega_{fi}-\omega)t} - 1}{\omega_{fi} - \omega}

$$


**On exact resonance** ($\omega = \omega_{fi}$), the numerator and denominator both vanish. Using L'H√īpital or direct substitution:

$$


\lim_{\omega \to \omega_{fi}} \frac{\mathrm{e}^{\mathrm{i}(\omega_{fi}-\omega)t} - 1}{\omega_{fi} - \omega} = \lim_{\delta \to 0} \frac{\mathrm{e}^{-\mathrm{i}\delta t} - 1}{-\delta} = it \mathrm{e}^{-\mathrm{i}\omega_{fi}t/2}\sin(\omega_{fi}t/2) / (\omega_{fi}t/2)

$$


Actually, on exact resonance:

$$


\mathcal{A}_{i\to f}(t) = -\mathrm{i}\frac{V_0}{2\hbar} t + O((\omega - \omega_{fi})^2)

$$


Wait, let me recalculate more carefully. On resonance:

$$


\int_0^t \mathrm{e}^{\mathrm{i}\Delta\omega t'} dt' \Big|_{\Delta\omega = 0} = t

$$


So:

$$


\mathcal{A}_{i\to f}(t) = -\frac{\mathrm{i} V_0 t}{2\hbar}

$$


But this grows without bound, which is unphysical (violates probability conservation). The issue is that we must treat near-resonance more carefully using the Rabi frequency. The full solution (without the first-order approximation) gives sinusoidal oscillations.

**Exact result from solving the two-level system directly**: For a two-level atom in a resonant harmonic field, the population oscillates as:

$$


P_{i\to f}(t) = \sin^2\left(\frac{\Omega_R t}{2}\right)

$$


where $\Omega_R = |V_{fi}|/\hbar$ (the Rabi frequency).

:::

:::{admonition} Stimulated Emission vs. Absorption
:class: note

In the presence of an oscillating field $V(t) \propto \cos(\omega t)$:

- **Absorption** ($i \to f$ with $E_f > E_i$): The system absorbs a photon and transitions upward. Resonance: $\hbar\omega = E_f - E_i$. This is the first term in the rotating-wave expansion.

- **Stimulated emission** ($i \to f$ with $E_f < E_i$): The system emits a photon stimulated by the oscillating field and transitions downward. Resonance: $\hbar\omega = E_i - E_f$. This corresponds to the second term.

Perturbation theory treats both transitions on equal footing‚ÄĒthey arise as different contributions in the expansion. The distinction (absorption vs. emission) is determined by the resonance condition and the sign of the energy difference. Both are deterministic (unlike spontaneous emission, which occurs without external driving).



:::{admonition} Discussion: Rotating Wave Validity
:class: dropdown warning

The rotating wave approximation (RWA) drops the rapidly oscillating counter-rotating term (frequency $\sim 2\omega$ from the second integral). This is justified when the Rabi frequency is much smaller than the driving frequency:

$$


\frac{|V_{fi}|}{\hbar} \ll \omega_{fi}

$$


- **What happens off-resonance?** When $\omega \neq \omega_{fi}$, the resonant term is no longer truly resonant. How does detuning $\Delta = \omega - \omega_{fi}$ affect RWA validity? Derive the condition on $\Delta$ for the approximation to remain valid.

- **Quantifying RWA error:** Estimate the transition amplitude from the discarded counter-rotating term when on resonance. How does the error scale with $\Omega_R / \omega$?

- **Strong driving regime:** When the Rabi frequency becomes comparable to the driving frequency ($\Omega_R \sim \omega$), what effects from counter-rotating terms become observable?

#### The Adiabatic Theorem Statement

Consider a **time-dependent Hamiltonian** $H(t)$ that depends on a parameter (or parameters) that evolve slowly. More precisely:

$$


H(t) = H_0(\lambda(t))

$$


where $\lambda(t)$ changes on a **time scale** $T$ that is much longer than $\hbar / (\Delta E)$, where $\Delta E$ is the typical energy scale (gap between levels).

:::{admonition} Adiabatic Theorem
:class: important If the system starts in an eigenstate $|n(0)\rangle$ of $H(0)$, it will remain in the **instantaneous eigenstate** $|n(t)\rangle$ of $H(t)$ (up to phase factors) throughout the evolution, provided the adiabatic condition is satisfied.

$$


|\psi(t)\rangle \approx \mathrm{e}^{\mathrm{i}\gamma_n(t)} \mathrm{e}^{-\mathrm{i}\phi_n(t)} |n(t)\rangle

$$ (eq-5c-adiabatic-state)

where:
- $\phi_n(t) = \frac{1}{\hbar} \int_0^t E_n(t') \mathrm{d}t\'$ is the **dynamical phase** (standard quantum phase acquired from time evolution).
- $\gamma_n(t)$ is the **Berry phase** (geometric phase)‚ÄĒa topological quantity depending on the path in parameter space.

#### Derivation from Perturbation Theory

The adiabatic theorem can be derived rigorously from time-dependent perturbation theory. The key insight: slowly varying parameters suppress transitions to other levels.

Consider the Hamiltonian in the **adiabatic basis** (the basis of instantaneous eigenstates $|n(t)\rangle$). The equation of motion in this basis contains **transition matrix elements**:

$$


\langle m(t) | \mathrm{i}\hbar \frac{\mathrm{d}}{\mathrm{d}t} | n(t) \rangle = (E_n - E_m) \delta_{mn} + \mathrm{i}\hbar \langle m | \frac{\partial}{\partial t} | n \rangle

$$


The second term (the "non-adiabatic coupling") is:

$$


\langle m | \frac{\partial}{\partial t} | n \rangle = \dot{\lambda}(t) \langle m | \frac{\partial}{\partial \lambda} | n \rangle

$$


**For slow changes** ($\dot{\lambda}$ small), this coupling is small, and transitions to other levels are suppressed. The adiabatic condition is:

$$


\left| \dot{\lambda}(t) \langle m | \frac{\partial}{\partial\lambda} | n \rangle \right| \ll |E_m(t) - E_n(t)| / \hbar

$$ (eq-5c-adiab-condition)

When this holds, transitions are negligible, and the system remains in state $|n(t)\rangle$ (up to phases).

:::{admonition} Derivation details
:class: dropdown note

Transform to the adiabatic basis by writing:

$$


|\psi(t)\rangle = \sum_n c_n(t) \mathrm{e}^{\mathrm{i}\gamma_n(t)} \mathrm{e}^{-\mathrm{i}\phi_n(t)} |n(t)\rangle

$$


where $\phi_n(t) = \int_0^t E_n(t') \mathrm{d}t\' / \hbar$ and $\gamma_n(t) = \mathrm{i} \int_0^t \langle n | \partial_t | n \rangle dt'$.

Substitute into the Schr√∂dinger equation $\mathrm{i}\hbar \partial_t |\psi\rangle = H |\psi\rangle$ and project onto $|m(t)\rangle$. The equation of motion for $c_m(t)$ is:

$$


\dot{c}_m(t) = -\dot{\lambda}(t) \sum_n c_n(t) \frac{\langle m | \partial_\lambda H | n \rangle}{E_n - E_m} \mathrm{e}^{\mathrm{i}(\gamma_m - \gamma_n)} \mathrm{e}^{-\mathrm{i}(\phi_m - \phi_n)}

$$


**Adiabatic approximation**: If $|\dot{\lambda}| / (\Delta E / \hbar) \ll 1$, then the coupling terms are small, and $c_n(t) \approx \delta_{n, n_0}$ (the coefficient remains constant for the initial state $n_0$). Thus the state remains $|n_0(t)\rangle$ to leading order.

:::

#### The Berry Phase (Geometric Phase)

The **Berry phase** $\gamma_n(t)$ arises from the **geometry of the path** traced in parameter space:

$$


\gamma_n(t) = \mathrm{i} \int_0^t \langle n(t') | \frac{\mathrm{d}}{\mathrm{d}t'} | n(t') \rangle dt'

$$ (eq-5c-berry-phase-integral)

**Key properties**:

1. **Geometric (not dynamical)**: Depends only on the path $\lambda(t)$ in parameter space, not on the speed of traversal or the time taken.

2. **Gauge-dependent**: The phase is gauge-dependent‚ÄĒit changes if we redefine $|n(t)\rangle \to \mathrm{e}^{\mathrm{i}\chi(t)} |n(t)\rangle$. However, for a **closed loop** (cyclic adiabatic evolution), the total Berry phase is **gauge-invariant**.

:::{admonition} Discussion: Gauge vs. Observability
:class: dropdown warning

The Berry phase $\gamma_n(t) = \mathrm{i} \int_0^t \langle n | \partial_t n \rangle dt'$ is explicitly gauge-dependent. Redefining $|n(t)\rangle \to \mathrm{e}^{\mathrm{i}\chi(t)} |n(t)\rangle$ shifts the Berry phase by $\chi(T) - \chi(0)$. Yet physicists insist it is "physical" and observable.

- **Open vs. closed paths:** Why is gauge freedom *not* a problem for measurements on open paths? What gauge-invariant quantity can you measure?

- **Cyclic evolution:** Show that returning to the original parameter point yields a gauge-invariant phase shift $\gamma_n^{\text{cyclic}}$. Why does closure remove the gauge ambiguity?

- **Experimental observation:** In the Aharonov-Bohm or spin Hall effect, how do interfering paths conspire to make the Berry phase observable despite individual gauge-dependent phases?

:::



3. **Topological (for closed loops)**: For cyclic evolution $\lambda(T) = \lambda(0)$:

$$


\gamma_n^{\text{cyclic}} = i \oint \langle n(\boldsymbol{R}) | \nabla_\boldsymbol{R} n(\boldsymbol{R}) \rangle \cdot d\boldsymbol{R}

$$


This is related to the **Berry curvature** and is a topological invariant (Chern number) depending only on the topology of the closed path, not its parametrization.

4. **Observable**: The Berry phase manifests as a relative phase shift between regions of space and can be observed via interference experiments (e.g., Aharonov-Bohm effect for non-abelian gauge fields, spin Hall effect).

**Further details**: The full geometric theory of Berry phases, including the Berry connection, curvature, and topological classification, is developed in **¬ß4.5.1 Geometric Phase**. That section shows the general framework for non-cyclic paths and reveals how topology constrains quantum evolution.

#### Example: Spin in Rotating Magnetic Field

A spin-1/2 in a magnetic field $\boldsymbol{B}(t) = B(\cos\omega t, \sin\omega t, 0)$ (precessing in the $xy$-plane) follows the field adiabatically if the field rotates slowly. In the rotating frame (adiabatic limit), the spin precesses around the field direction. After a complete rotation, the spin acquires a Berry phase:

$$


\gamma = \frac{\pi}{2} \Omega \quad (\text{for a } 2\pi \text{ rotation angle})

$$


This phase is observable and represents a purely geometric effect‚ÄĒit depends on the path traced, not the time taken.

### Application 3: Green's Functions and Scattering

#### Definition

The **Green's function** (propagator or resolvent) is a fundamental object encoding all spectral information:

$$


\boxed{G^\pm(E) = \frac{1}{E - H \pm \mathrm{i}\epsilon}}

$$ (eq-5c-green-full)

where $\epsilon \to 0^+$ and:
- $G^+$ (retarded): Describes outgoing boundary conditions (effect follows cause).
- $G^-$ (advanced): Describes incoming boundary conditions.

In time domain:

$$


G^+(\boldsymbol{r}, \boldsymbol{r}'; E) = -\frac{\mathrm{i}}{\hbar} \int_0^\infty dt \, \mathrm{e}^{\mathrm{i}Et/\hbar} \langle \boldsymbol{r} | \mathrm{e}^{-\mathrm{i}Ht/\hbar} | \boldsymbol{r}' \rangle

$$


#### Physical Interpretation

The pole structure of $G(E)$ reveals the quantum spectrum:

- **Bound states**: Poles at $E < 0$ (below continuum threshold) with infinitesimal imaginary part.
- **Resonances**: Poles in the lower half-plane at $E_R - i\Gamma/2$, indicating decaying states with lifetime $\tau = \hbar/\Gamma$.
- **Scattering states (continuum)**: Branch cuts along the real axis.

#### Lippmann-Schwinger Equation

For a system with Hamiltonian $H = H_0 + V$, the full Green's function satisfies a self-consistent equation:

$$


\boxed{G^\pm(E) = G_0^\pm(E) + G_0^\pm(E) V G^\pm(E)}

$$ (eq-5c-lippmann-schwinger)

where $G_0^\pm(E) = 1/(E - H_0 \pm \mathrm{i}\epsilon)$ is the free (unperturbed) Green's function.

**Iterating this equation** generates the **Born series**:

$$


\boxed{G^\pm(E) = G_0^\pm + G_0^\pm V G_0^\pm + G_0^\pm V G_0^\pm V G_0^\pm + \ldots}

$$ (eq-5c-born-series)

Each term represents an additional scattering event from $V$.

:::{admonition} Derivation
:class: dropdown note

Start with:

$$


(E - H_0 - V) G^\pm(E) = 1

$$


Rewrite as:

$$


(E - H_0) G^\pm(E) = 1 + V G^\pm(E)

$$


Multiply both sides by $G_0^\pm(E) = 1/(E - H_0 \pm \mathrm{i}\epsilon)$:

$$


G^\pm(E) = G_0^\pm(E) + G_0^\pm(E) V G^\pm(E)

$$


This is the Lippmann-Schwinger equation. Iterate:

$$


G^\pm = G_0^\pm + G_0^\pm V G_0^\pm + G_0^\pm V G_0^\pm V G_0^\pm + \ldots

$$


:::

#### Scattering T-Matrix and Born Approximation

The **T-matrix** (transition operator) is defined by:

$$


G^\pm(E) = G_0^\pm(E) + G_0^\pm(E) T^\pm(E) G_0^\pm(E)

$$


The **Born approximation** (first-order scattering) truncates at:

$$


T^{(1)}(E) \approx V

$$


The **scattering amplitude** is proportional to the T-matrix element:

$$


f(\theta) \propto \langle \boldsymbol{k}_f | T(E) | \boldsymbol{k}_i \rangle

$$


where $|\boldsymbol{k}_i\rangle$ and $|\boldsymbol{k}_f\rangle$ are momentum eigenstates.

### Application 4: Feynman Diagrams

#### Graphical Representation of Perturbation Series

**Feynman diagrams** provide a visual encoding of the perturbative expansion. Each diagram represents a term in the series expansion.

**Basic elements**:
- **External lines**: Initial and final particle states (entering/leaving).
- **Internal lines**: Intermediate states (virtual particles), propagating via the free Green's function $G_0^\pm$.
- **Vertices**: Interaction points from the perturbation $V$.

**Feynman rules** for a given perturbation theory:
1. Draw all topologically distinct diagrams with $n$ vertices (corresponding to order $\lambda^n$).
2. For each diagram:
   - Each internal line contributes a factor $G_0^\pm(E)$ (propagator).
   - Each vertex contributes a matrix element from $V$.
   - Integrate over all intermediate energies and internal momenta.
3. Include combinatorial factors (symmetry factors, signs for fermions).

#### Example: Second-Order Diagram

For a transition $i \to n \to f$ via an intermediate state, the diagram is:

$$


\text{Diagram:} \quad |i\rangle \to |n\rangle \to |f\rangle

$$


This contributes:

$$


\text{Contribution} = \frac{\langle f|V|n\rangle \langle n|V|i\rangle}{E_i^{(0)} - E_n^{(0)} + \mathrm{i}\epsilon}

$$


The denominator encodes the energy difference; the $\mathrm{i}\epsilon$ term ensures proper causality (the intermediate state is "virtual" and off-shell).

#### Advantages

Feynman diagrams provide:
- **Intuitive pictures** of complex quantum processes.
- **Systematic enumeration** of all perturbative contributions at a given order.
- **Error checking**: Easy to identify which topologies have been included or missed.
- **Path to field theory**: In QED, QCD, and other field theories, Feynman diagrams are the primary computational tool.
- **Physical insight**: Virtual particles, momentum conservation, energy conservation all have graphical representations.

### Summary

- **Harmonic driving**: Creates **Rabi oscillations** at frequency $\Omega_R = |V|/\hbar$. Both absorption and stimulated emission are treated on equal footing; resonance determines which dominates.
- **Adiabatic evolution**: Slow parameter changes keep the system in the **instantaneous eigenstate**, acquiring a **Berry phase** that is geometric (path-dependent, not speed-dependent) and topological for closed loops.
- **Green's functions**: Encode the quantum spectrum (bound states, resonances, continuum). Satisfy self-consistent **Lippmann-Schwinger equation**; Born series expands in powers of perturbation.
- **Feynman diagrams**: Visual representation of every term in the perturbative expansion. Vertices, propagators, and integration rules provide systematic calculational framework.
- **Common themes across applications**: Time ordering, energy conservation, resonance conditions, systematic perturbative expansion, and quantum coherence.

(c) Explain why this phase is **geometric**: it depends on the solid angle subtended by the path traced by $\boldsymbol{B}(t)$ in parameter space, not on the rotation rate $\Omega$ or the time $T$.

(d) Is the Berry phase observable? How could you design an interference experiment to detect it?

**6. Green's function poles and spectrum**

The retarded Green's function for a quantum system is:

$$


G^+(E) = \frac{1}{E - H + \mathrm{i}\epsilon}

$$


The poles of $G^+(E)$ reveal the quantum spectrum.

(a) For a **bound state** with energy $E_n < 0$, the Green's function has a pole at $E = E_n$ (on the real axis). Show that the residue at this pole is proportional to the projection onto the bound state: $|n\rangle\langle n|$.

(b) A **resonance** corresponds to a state that decays. Its energy is complex: $E_R - i\Gamma/2$, where $\Gamma$ is the decay rate (half-width). In the complex $E$-plane, is this pole in the upper or lower half-plane? Why does the small imaginary part describe exponential decay?

(c) For a potential $V(\boldsymbol{r}) = -V_0 \mathrm{e}^{-r/a}$ (exponential well), sketch the analytic structure of $G^+(E)$ in the complex $E$-plane. Include: bound state poles, continuum branch cut, and an example resonance.

(d) How would you extract the bound state energy and the scattering phase shifts from the Green's function?

**7. Born approximation and scattering length**

Consider low-energy scattering of a particle from a short-range potential $V(\boldsymbol{r})$ with range $a$. In the Born approximation, the scattering amplitude is:

$$


f(\theta) \approx -\frac{m}{2\pi\hbar^2} \int d^3 r \, \mathrm{e}^{\mathrm{i}\boldsymbol{q}\cdot\boldsymbol{r}} V(\boldsymbol{r})

$$


where $\boldsymbol{q} = \boldsymbol{k}_i - \boldsymbol{k}_f$ is the momentum transfer.

(a) For a **spherically symmetric** potential, the forward scattering amplitude ($\theta = 0$, $\boldsymbol{q} = 0$) is:

$$


f(0) = -\frac{m}{2\pi\hbar^2} \int d^3 r \, V(\boldsymbol{r}) = -\frac{m}{2\pi\hbar^2} \langle V \rangle

$$


The **scattering length** $a_s$ is related to forward scattering by $f(0) = -a_s$. Show that $a_s = \frac{m}{2\pi\hbar^2} \int d^3 r \, V(\boldsymbol{r})$.

(b) For a square-well potential $V(r) = -V_0$ for $r < a$ and $V(r) = 0$ for $r > a$, calculate the scattering length in the Born approximation.

(c) When is the Born approximation valid? State the condition in terms of $V_0$, $a$, and the particle energy $E$.

(d) For attractive potentials ($V_0 > 0$), the Born approximation gives $a_s > 0$. Can the exact (non-perturbative) solution have $a_s < 0$? What does this imply?

**8. Feynman diagrams for fourth-order scattering**

Consider a particle scattering off a potential $V$. The Born series gives the scattering amplitude as a sum over diagrams:

- **First order** (Born approximation): One vertex, no intermediate propagators. Contributes $T^{(1)} \sim V$.
- **Second order**: Two vertices, one intermediate propagator. Contributes $T^{(2)} \sim V G_0 V$.

(c) In fourth order, how many distinct Feynman topologies (distinct orderings of vertices and propagators) contribute? Why is combinatorial care needed to avoid double-counting?
