## Lecture Notes

Classical information theory measures uncertainty through Shannon entropy; quantum mechanics demands its own generalization. Von Neumann entropy quantifies the mixedness of a quantum state, ranging from zero for perfect knowledge to a maximum for complete ignorance. This single number encodes how much pure quantum coherence remains, serving as the quantum analog of 'how much information we have lost.'

:::{admonition} Definition: Von Neumann Entropy
:class: important

For a density matrix $\rho$, the **von Neumann entropy** quantifies the quantum information (or uncertainty) in the state:

$$


S(\rho) = -\text{Tr}(\rho \ln \rho)

$$ (eq-entropy-vn-def)

In the eigenbasis of $\rho$ with eigenvalues $\{\lambda_i\}$, this simplifies to:

$$


S(\rho) = -\sum_i \lambda_i \ln \lambda_i

$$ (eq-entropy-vn-eigenbasis)

where we use the convention $0 \ln 0 = 0$ (taking the limit as $x \to 0^+$: $x \ln x \to 0$).
:::

:::{admonition} Physical Interpretation
:class: note


- **Pure state** ($\rho = |\psi\rangle\langle\psi|$): $S(\rho) = 0$ (complete certainty, no entropy)
- **Maximally mixed** ($\rho = I/d$): $S(\rho) = \ln d$ (maximum entropy for dimension $d$)
- **General mixed state**: $0 \leq S(\rho) \leq \ln d$

**Units**: We use natural logarithm, giving entropy in **nats**. Base-2 logarithm gives **bits** (1 nat $\approx 1.44$ bits).

:::{admonition} Derivation: Von Neumann Entropy
:class: dropdown note

**Construction of the formula:**

Start with the von Neumann operator:

$$


S(\rho) = -\text{Tr}(\rho \ln \rho)

$$


To evaluate this, work in the eigenbasis where $\rho = \sum_i \lambda_i |u_i\rangle\langle u_i|$.

Then $\ln \rho = \sum_i (\ln \lambda_i) |u_i\rangle\langle u_i|$, so:

$$


\rho \ln \rho = \sum_i \lambda_i (\ln \lambda_i) |u_i\rangle\langle u_i|

$$


Taking the trace:

$$


\text{Tr}(\rho \ln \rho) = \sum_i \lambda_i \ln \lambda_i

$$


Therefore:

$$


S(\rho) = -\sum_i \lambda_i \ln \lambda_i

$$


This is the **spectral formula** and is independent of basis choice (eigenvalues are invariant).

**Limit convention**: For eigenvalues $\lambda_i = 0$, the term $\lambda_i \ln \lambda_i$ is defined to vanish:

$$


\lim_{x \to 0^+} x \ln x = 0

$$


This makes physical sense: a zero eigenvalue contributes no entropy (it's not in the support of the state).

:::
:::

###### Classical vs. Quantum Entropy

**Classical Shannon entropy** for a probability distribution $\{p_i\}$:

$$


H(p) = -\sum_i p_i \ln p_i

$$


**Quantum von Neumann entropy** generalizes this: the eigenvalues of $\rho$ play the role of probabilities, but $\rho$ encodes more than classical probabilities (coherences, entanglement).

**Key difference**: Quantum entropy can increase or decrease under local operations (unlike classical entropy, which cannot decrease under any operation).

### Fundamental Examples

#### Pure State

For $\rho = |\psi\rangle\langle\psi|$ (eigenvalues: 1 and 0's):

$$


S(\rho) = -1 \cdot \ln 1 - 0 \cdot \ln 0 = 0

$$ (eq-entropy-pure)

**No uncertainty** about the state; entropy is zero.

#### Maximally Mixed State (d-dimensional)

For $\rho = I/d$ (eigenvalues: all $1/d$):

$$


S(\rho) = -d \cdot \frac{1}{d} \ln\frac{1}{d} = \ln d

$$ (eq-entropy-maxmix)

For a qubit ($d=2$): $S(I/2) = \ln 2 \approx 0.693$ bits.

**Maximum uncertainty** for a $d$-dimensional system.

#### Mixed Qubit (Example)

Eigenvalues $\lambda_1 = 0.7, \lambda_2 = 0.3$:

$$


S(\rho) = -0.7 \ln(0.7) - 0.3 \ln(0.3) \approx 0.610 + 1.204 \approx 1.204\text{ nats} \approx 0.610\text{ bits}

$$ (eq-entropy-mixed-example)

Intermediate between pure (0) and maximally mixed ($\ln 2 \approx 0.693$ bits).

### Properties of Von Neumann Entropy

#### 1. Boundedness

:::{admonition} Entropy Bounds
:class: important

For a density matrix on $d$-dimensional Hilbert space:

$$


0 \leq S(\rho) \leq \ln d

$$ (eq-entropy-bound)

- Lower bound: $S(\rho) = 0$ ⟺ $\rho$ is pure (rank 1).
- Upper bound: $S(\rho) = \ln d$ ⟺ $\rho = I/d$ (maximally mixed, full rank).
:::

#### 2. Concavity

For any $0 \leq p \leq 1$ and density matrices $\rho_1, \rho_2$:

$$


S(p\rho_1 + (1-p)\rho_2) \geq p S(\rho_1) + (1-p)S(\rho_2)

$$ (eq-entropy-concave)

**Interpretation**: Entropy of a mixture is **at least** the average entropy of the components. Mixing increases entropy (or keeps it same).

**Consequence**: If you don't know which state you have (ensemble), entropy is higher than if you knew which one it was.

#### 3. Additivity (For Independent Systems)

If $\rho_{AB} = \rho_A \otimes \rho_B$ (product state, no entanglement):

$$


S(\rho_A \otimes \rho_B) = S(\rho_A) + S(\rho_B)

$$ (eq-entropy-additive)

Entropies add when systems are uncorrelated.

#### 4. Subadditivity (Araki–Lieb Inequality)

:::{admonition} Subadditivity of Entropy
:class: important

For any bipartite state $\rho_{AB}$:

$$


S(\rho_{AB}) \leq S(\rho_A) + S(\rho_B)

$$ (eq-entropy-subadditive)

**Interpretation**: Joint entropy is **at most** the sum of marginal entropies. Entanglement reduces the total entropy.

**Equality** ⟺ no entanglement: $\rho_{AB} = \rho_A \otimes \rho_B$.

**For pure states**: $S(\rho_{AB}) = 0$, so $S(\rho_A) + S(\rho_B) \geq 0$ (always true, but becomes $S(\rho_A) = S(\rho_B) = E_{\text{ent}} > 0$ if entangled).
:::

#### 5. Strong Subadditivity

For a tripartite system $ABC$:

$$


S(\rho_{AB}) + S(\rho_{BC}) \geq S(\rho_B) + S(\rho_{ABC})

$$ (eq-entropy-strong-subadditive)

This is the cornerstone of the **second law of thermodynamics** in quantum information.

### Maximum Entropy Principle

**Principle**: Among all density matrices consistent with certain constraints (e.g., given expectation values), the one with **maximum entropy** represents maximum ignorance and is the most natural.

#### Thermal State (Example)

At thermal equilibrium, a system reaches the state that maximizes entropy subject to fixed average energy $\langle H \rangle = E$.

:::{admonition} Thermal State and Maximum Entropy
:class: important

Using **Lagrange multipliers** (with energy constraint and normalization):

$$


\rho_\text{thermal} = \frac{\mathrm{e}^{-\beta H}}{\text{Tr}(\mathrm{e}^{-\beta H})} = \frac{\mathrm{e}^{-\beta H}}{Z}

$$ (eq-entropy-thermal-state)

where $\beta = 1/(k_B T)$ (inverse temperature) and $Z = \text{Tr}(\mathrm{e}^{-\beta H})$ (partition function).

This maximizes von Neumann entropy for fixed average energy: $S(\rho_\text{thermal})$ is maximum among all $\rho$ with $\langle H \rangle = E$.
:::

#### Maximally Mixed State

Unconstrained maximum entropy: $\rho = I/d$, with $S_\text{max} = \ln d$.

### Von Neumann Entropy and Thermodynamic Entropy

**Boltzmann–Gibbs relation** (fundamental in statistical mechanics):

$$


S_\text{thermo} = k_B \cdot S_\text{von Neumann}(\rho)

$$ (eq-entropy-thermodynamic-connection)

where:
- $S_\text{thermo}$ = thermodynamic entropy (extensive, increases with system size).
- $S_\text{von Neumann}(\rho)$ = quantum information entropy.
- $k_B$ = Boltzmann constant ≈ $1.38 \times 10^{-23}$ J/K.

**Interpretation**: Quantum information entropy is the **microscopic origin** of thermodynamic entropy. Uncertainty in microscopic state manifests as macroscopic irreversibility.

### Rényi Entropy

:::{admonition} Definition: Rényi Entropy Family
:class: important

A family of entropies parameterized by $\alpha \in (0, \infty) \setminus \{1\}$:

$$


S_\alpha(\rho) = \frac{1}{1-\alpha} \ln \text{Tr}(\rho^\alpha)

$$ (eq-entropy-renyi)

**Limit**: $\lim_{\alpha \to 1} S_\alpha(\rho) = S(\rho)$ (von Neumann entropy).
:::

#### Common Cases

**$\alpha = 0$ (Hartley entropy)**:

$$


S_0(\rho) = \ln(\text{rank}(\rho))

$$


Counts number of nonzero eigenvalues.

**$\alpha = 1/2$ (Collision entropy variant)**:

$$


S_{1/2}(\rho) = -2\ln \text{Tr}(\rho^{1/2})

$$


**$\alpha = 2$ (Rényi-2, collision entropy)**:

$$


S_2(\rho) = -\ln \text{Tr}(\rho^2) = -\ln(\text{Purity})

$$ (eq-entropy-renyi-2)

Related to purity: $\text{Tr}(\rho^2) = \mathrm{e}^{-S_2(\rho)}$.

**$\alpha \to \infty$ (Min-entropy)**:

$$


S_\infty(\rho) = -\ln(\lambda_\text{max})

$$


where $\lambda_\text{max}$ is the largest eigenvalue.

#### Properties

1. **Monotonicity**: $S_\alpha(\rho) \geq S_{\alpha'}(\rho)$ for $\alpha \leq \alpha'$ (entropy decreases with order).

2. **Bounds von Neumann**: $S_\infty(\rho) \leq S_1(\rho) \leq S_0(\rho)$.

3. **Operational meaning**: Different $\alpha$ values capture different aspects of information:
   - $S_0$: Worst-case (max number of distinguishable states).
   - $S_1$: Average-case (von Neumann).
   - $S_2$: Accessible without full tomography (measure $\text{Tr}(\rho^2)$ via correlations).
   - $S_\infty$: Best-case (min-entropy; security-relevant).

#### Rényi-2 Experimental Accessibility

For identical independent measurements, the probability of getting the same outcome twice is:

$$


P(\text{same}) = \text{Tr}(\rho^2)

$$


Thus $S_2(\rho) = -\ln P(\text{same})$ is **directly measurable** without reconstructing full $\rho$ (no need for full state tomography).

### Conditional Entropy

For a bipartite state $\rho_{AB}$, the **conditional entropy** of A given B is:

$$


S(A|B) = S(\rho_{AB}) - S(\rho_B)

$$ (eq-entropy-conditional)

**Interpretation**: Average entropy of A after measuring B and learning the outcome (on average).

**Properties**:
- $S(A|B) \leq S(A)$ (knowing B cannot increase A's entropy).
- $S(A|B) \geq S(A) - S(B)$ (Araki–Lieb inequality).
- For entangled pure states: $S(A|B) = -S(A) = -S(B)$ (negative, indicating strong correlation).

### Mutual Information and Entanglement Entropy

**Mutual information** (total correlations):

$$


I(A : B) = S(A) + S(B) - S(AB)

$$


For a **pure bipartite state** where $S(AB) = 0$:

$$


I(A : B) = S(A) + S(B) = 2 S(A) = 2 E_{\text{ent}}

$$ (eq-entropy-mutual-ent)

Thus **entanglement entropy** is half the mutual information for pure bipartite states.

:::{admonition} Entropy of subsystem can exceed total
:class: attention
For an entangled pure bipartite state, $S(\rho_{AB}) = 0$ but $S(\rho_A) > 0$. The subsystem has more entropy than the whole—a uniquely quantum phenomenon with no classical analogue. This is possible because quantum correlations (entanglement) are not captured by the marginals.
:::

### Summary

- **Von Neumann entropy** $S(\rho) = -\text{Tr}(\rho \ln \rho) = -\sum_i \lambda_i \ln \lambda_i$: Quantifies ignorance/mixedness; zero for pure, $\ln d$ for maximally mixed.
- **Boundedness**: $0 \leq S(\rho) \leq \ln d$.
- **Concavity**: Averaging increases entropy (or keeps same).
- **Additivity**: For independent systems, $S(\rho_A \otimes \rho_B) = S(\rho_A) + S(\rho_B)$.
- **Subadditivity**: $S(\rho_{AB}) \leq S(\rho_A) + S(\rho_B)$; entanglement reduces total entropy.
- **Maximum entropy principle**: Thermal state $\rho = \mathrm{e}^{-\beta H}/Z$ maximizes entropy at fixed average energy.
- **Thermodynamic entropy**: $S_\text{thermo} = k_B S_\text{von Neumann}$; quantum uncertainty is origin of irreversibility.
- **Rényi entropy** $S_\alpha(\rho) = \frac{1}{1-\alpha}\ln\text{Tr}(\rho^\alpha)$: Family generalizing von Neumann; $S_2$ experimentally accessible.
- **Conditional entropy**: $S(A|B) = S(AB) - S(B)$; measures A's entropy after learning B.
- **Mutual information**: $I(A:B) = S(A) + S(B) - S(AB)$; for pure bipartite, $= 2E_{\text{ent}}$.
