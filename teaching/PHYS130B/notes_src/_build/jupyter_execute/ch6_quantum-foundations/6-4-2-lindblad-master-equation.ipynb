## Lecture Notes

For an isolated system, the Schrödinger equation perfectly captures quantum evolution. But real systems interact with their surroundings—heat baths, photon fields, uncontrolled degrees of freedom. The Lindblad master equation describes how open systems evolve, accounting for both unitary dynamics and the irreversible effects of environmental coupling. It is the most general form of Markovian (memoryless) evolution compatible with quantum mechanics.

### Master Equation: General Form



:::{admonition} Discussion: When Is the Markovian Approximation Valid?
:class: dropdown warning

The Lindblad equation assumes the system and environment are uncorrelated at each instant—a Markovian approximation. But in reality, the environment can have memory. How do we know when Markovian is okay?

- The Markovian approximation breaks down if the environment correlation time is comparable to the system timescale. How do we compute that correlation time from first principles?
- Some systems show "non-Markovian revivals"—the system state temporarily recovers coherence it lost earlier. Does this invalidate the Lindblad equation, or is it simply outside its regime of applicability?
- If we keep the Lindblad form but allow time-dependent $\gamma_i(t)$, have we fixed the approximation, or just hidden the non-Markovianity in the time dependence?
:::


For an open system coupled to environment (bath/reservoir), the **master equation** describes the reduced density matrix $\rho_S$ (system only):

$$


\frac{\mathrm{d}\rho_S}{dt} = \mathcal{L}(\rho_S)

$$ (eq-lindblad-master-general)

where $\mathcal{L}$ is the **Liouvillian superoperator** (linear map on density matrices).

We derive the Lindblad equation from first principles:

:::{admonition} Derivation: Lindblad Master Equation from System-Environment
:class: dropdown note

Start with total Hamiltonian:

$$


H_\text{total} = H_S \otimes I_E + I_S \otimes H_E + H_\text{int}

$$


The full system+environment evolves unitarily. Tracing out the environment (and assuming **Markovian approximation**: short environment memory):

$$


\frac{\mathrm{d}\rho_S}{dt} = -\frac{\mathrm{i}}{\hbar}[H_S, \rho_S] + \mathcal{D}(\rho_S)

$$ (eq-lindblad-master-decomposed)

where:
- First term: unitary evolution under system Hamiltonian.
- Second term: **dissipation superoperator** describing decoherence.
:::
### The Lindblad Master Equation (GKSL Form)

:::{admonition} Lindblad Master Equation (GKSL Form)
:class: important

The most general form preserving trace and positivity is the **Lindblad equation**:

$$


\frac{\mathrm{d}\rho}{dt} = -\frac{\mathrm{i}}{\hbar}[H, \rho] + \sum_k \gamma_k \left( L_k \rho L_k^\dagger - \frac{1}{2}\{L_k^\dagger L_k, \rho\} \right)

$$ (eq-lindblad-equation)

where:
- $H$ = system Hamiltonian (Hermitian).
- $L_k$ = **Lindblad operators** (or **jump operators**); arbitrary operators (generally non-Hermitian).
- $\gamma_k \geq 0$ = **decay rates** (non-negative).
- $\{A, B\} = AB + BA$ = anticommutator.

Named: Lindblad (1976), Kossakowski, Sudarshan (GKSL form). This is the most general form of Markovian evolution compatible with complete positivity and trace preservation.

:::

### Physical Meaning of Terms

#### Unitary Evolution

$$


-\frac{\mathrm{i}}{\hbar}[H, \rho]

$$


Standard Schrödinger evolution of the isolated system.

#### Gain Term

$$


\gamma_k L_k \rho L_k^\dagger

$$


**Gain** of population by jumping via $L_k$. Represents an **inflow** of probability as system undergoes transition $L_k$.

#### Loss Term

$$


-\frac{1}{2}\gamma_k \{L_k^\dagger L_k, \rho\}

$$


**Loss** of population by jumping out of the current state via $L_k$. Ensures trace is conserved:

$$


\text{Tr}\left(-\frac{1}{2}\{L_k^\dagger L_k, \rho\}\right) = -\text{Tr}(L_k^\dagger L_k \rho) = -\text{Tr}(L_k \rho L_k^\dagger)

$$


Gain and loss balance to preserve $\text{Tr}(\rho) = 1$.

### Properties of the Lindblad Equation

:::{admonition} Key Properties of Lindblad Evolution
:class: important

The Lindblad master equation preserves the fundamental properties of quantum states:

**1. Trace Preservation**: $\frac{\mathrm{d}}{\mathrm{d}t}\text{Tr}(\rho) = 0$  
$\Rightarrow$ $\text{Tr}(\rho(t)) = 1$ for all $t$ (probability conservation)

**2. Hermiticity Preservation**: If $\rho(0)$ is Hermitian, then $\rho(t)$ remains Hermitian

**3. Positivity (Positive Semidefiniteness)**: If $\rho(0) \geq 0$, then $\rho(t) \geq 0$ for all $t$  
$\Rightarrow$ Evolution generates only valid quantum states

**4. CPTP (Completely Positive Trace-Preserving)**: The time evolution generates a CPTP map (quantum channel); mathematically equivalent to Kraus representation with operators $K_k = \sqrt{\gamma_k} L_k$.

These properties are necessary and sufficient for evolution to respect quantum mechanics; any master equation violating them is unphysical.

:::



**1. Trace Preservation:**

$$


\frac{\mathrm{d}}{\mathrm{d}t}\text{Tr}(\rho) = 0 \Rightarrow \text{Tr}(\rho(t)) = 1

$$ (eq-lindblad-trace-preserving)

**2. Hermiticity Preservation:**
If $\rho(0)$ is Hermitian, then $\rho(t)$ remains Hermitian.

**3. Positive Semidefiniteness (Positivity-Preserving):**
If $\rho(0) \geq 0$, then $\rho(t) \geq 0$ for all $t$ (valid quantum state).

**4. CPTP (Completely Positive Trace-Preserving):**
The time evolution generates a CPTP map (quantum channel); formally equivalent to Kraus representation.

### Common Lindblad Operators

#### 1. Amplitude Damping: $L = \sqrt{\Gamma_1} \hat{\sigma}_- = \sqrt{\Gamma_1} |{\downarrow}\rangle\langle{\uparrow}|$

**Physical**: Excited state decays to ground state; energy dissipation at rate $\Gamma_1 = 1/T_1$.

**Master equation**:

$$


\frac{\mathrm{d}\rho}{dt} = \Gamma_1 \left( \hat{\sigma}_- \rho \hat{\sigma}_+ - \frac{1}{2}\{\hat{\sigma}_+ \hat{\sigma}_-, \rho\} \right)

$$ (eq-lindblad-amp-damping)

**Effect on populations**:

$$


\frac{\mathrm{d}\rho_{\uparrow\uparrow}}{dt} = -\Gamma_1 \rho_{\uparrow\uparrow}, \quad \frac{\mathrm{d}\rho_{\downarrow\downarrow}}{dt} = +\Gamma_1 \rho_{\uparrow\uparrow}

$$


Population decays from excited ($|{\uparrow}\rangle$) to ground ($|{\downarrow}\rangle$) at rate $\Gamma_1$.

**Effect on coherence**:

$$


\frac{\mathrm{d}\rho_{\uparrow\downarrow}}{dt} = -\frac{\Gamma_1}{2} \rho_{\uparrow\downarrow}

$$


Coherences decay at half the rate (because excited state population decreases).

#### 2. Phase Damping (Dephasing): $L = \sqrt{\Gamma_\phi} \hat{\sigma}^z$

**Physical**: Random phase fluctuations; coherences decay without energy loss at rate $\Gamma_\phi = 1/T_2$ (where $T_2 > T_1/2$).

**Master equation**:

$$


\frac{\mathrm{d}\rho}{dt} = \Gamma_\phi (\hat{\sigma}^z \rho \hat{\sigma}^z - \rho)

$$ (eq-lindblad-phase-damping)

**Effect**: Only off-diagonal (coherence) terms decay:

$$


\rho_{\uparrow\downarrow}(t) = \mathrm{e}^{-\Gamma_\phi t} \rho_{\uparrow\downarrow}(0)

$$


Diagonal (population) terms unchanged. Dephasing creates classical mixtures.

#### 3. Depolarizing: $L_x = \sqrt{\Gamma_d/3} \hat{\sigma}^x$, $L_y = \sqrt{\Gamma_d/3} \hat{\sigma}^y$, $L_z = \sqrt{\Gamma_d/3} \hat{\sigma}^z$

**Physical**: Errors on all three Pauli axes with equal probability; generic noise at rate $\Gamma_d$.

**Master equation**: Drives state toward maximally mixed $I/2$ (after time $\sim 1/\Gamma_d$).

### Steady States

A **steady state** $\rho_\infty$ satisfies:

$$


\mathcal{L}(\rho_\infty) = 0

$$ (eq-lindblad-steady-state)

i.e., $\frac{\mathrm{d}\rho_\infty}{dt} = 0$; the state does not evolve.

:::{admonition} Example: Amplitude Damping
:class: example


From $\frac{\mathrm{d}\rho_{\uparrow\uparrow}}{dt} = -\Gamma_1 \rho_{\uparrow\uparrow}$, steady state has $\rho_{\uparrow\uparrow}^\infty = 0$.

Thus: $\rho_\infty = |{\downarrow}\rangle\langle{\downarrow}|$ (ground state; all population decayed).

:::
:::{admonition} Example: Amplitude Damping + Pumping
:class: example


If system is pumped into excited state at rate $\Gamma_2$ (with operator $L_2 = \sqrt{\Gamma_2} \hat{\sigma}_+$):

$$


\frac{\mathrm{d}\rho_{\uparrow\uparrow}}{dt} = -\Gamma_1 \rho_{\uparrow\uparrow} + \Gamma_2 \rho_{\downarrow\downarrow}

$$


Steady state: balance between decay and pumping.

$$


\rho_{\uparrow\uparrow}^\infty = \frac{\Gamma_2}{\Gamma_1 + \Gamma_2}, \quad \rho_{\downarrow\downarrow}^\infty = \frac{\Gamma_1}{\Gamma_1 + \Gamma_2}

$$


:::
### Quantum Jump Picture (Stochastic Interpretation)

**Alternative viewpoint**: Lindblad equation is the **average** over many quantum trajectories.

**Trajectory picture**:
1. System evolves under modified Hamiltonian:

   $$


   H_\text{eff} = H - \frac{\mathrm{i}\hbar}{2} \sum_k \gamma_k L_k^\dagger L_k

   $$


   (non-Hermitian: evolution is non-unitary).

2. At random times (Poisson process), a "**jump**" occurs: $\rho \to L_k \rho L_k^\dagger / p_k$ with probability $p_k \propto \gamma_k$.

3. Averaging over all trajectories (all possible jump sequences) reproduces the Lindblad master equation.

**Insight**: Decoherence and dissipation arise from random quantum jumps induced by the environment; averaging erases the randomness, revealing the ensemble behavior.

### Quantum Regression Theorem

For a system governed by Lindblad equation with steady state $\rho_\infty$, the **two-time correlation** is:

$$


\langle A(t) B(0) \rangle = \text{Tr}(A \mathcal{E}^{(t)}(B \rho_\infty))

$$ (eq-lindblad-regression)

where $\mathcal{E}^{(t)}$ is the time-evolved Liouvillian (solution of master equation).

**Interpretation**: Correlations at different times obey the same master equation as populations. One can predict oscillations and decay of correlations from the same Lindblad equation.

### Solving the Lindblad Equation

**General case**: Lindblad equation is a system of coupled ODEs for matrix elements of $\rho$.

**Methods**:
1. **Analytical**: For simple systems (single qubit, 2-level system); exact solutions exist.
2. **Numerical integration**: RK4, implicit methods; QuTiP library (Python) widely used.
3. **Eigenbasis decomposition**: Diagonalize Liouvillian matrix; solutions are exponentials $\mathrm{e}^{\lambda_i t}$.

:::{admonition} Example: Decay from Excited State
:class: example


Initial condition: $\rho(0) = |{\uparrow}\rangle\langle{\uparrow}|$. Under amplitude damping:

$$


\rho(t) = \begin{pmatrix} (1-\mathrm{e}^{-\Gamma_1 t}) & 0 \\ 0 & \mathrm{e}^{-\Gamma_1 t} \end{pmatrix}

$$


Wait: Let me recalculate. $\rho_{\uparrow\uparrow}(t) = \mathrm{e}^{-\Gamma_1 t}$, so:

$$


\rho(t) = \begin{pmatrix} \mathrm{e}^{-\Gamma_1 t} & 0 \\ 0 & 1 - \mathrm{e}^{-\Gamma_1 t} \end{pmatrix}

$$ (eq-lindblad-decay-solution)

Excited state population decays exponentially; ground state population grows. As $t \to \infty$: $\rho(\infty) = |{\downarrow}\rangle\langle{\downarrow}|$ (ground state).

:::
:::{admonition} Lindblad is Markovian
:class: attention
The Lindblad master equation assumes Markovian (memoryless) dynamics: the environment correlation time is much shorter than the system timescale. For structured environments or strong coupling, non-Markovian generalizations are needed.
:::

### Summary

- **Master equation**: Differential equation for reduced density matrix of open system; describes irreversible evolution.
- **Lindblad form** $\frac{\mathrm{d}\rho}{dt} = -\frac{\mathrm{i}}{\hbar}[H,\rho] + \sum_k \gamma_k(L_k\rho L_k^\dagger - \frac{1}{2}\{L_k^\dagger L_k, \rho\})$: Most general Markovian master equation.
- **Lindblad operators** $L_k$: Represent quantum jumps/transitions; rates $\gamma_k \geq 0$.
- **CPTP**: Lindblad evolution is completely positive, trace-preserving; valid quantum dynamics.
- **Common operators**: Amplitude damping ($\hat{\sigma}_-$), phase damping ($\hat{\sigma}^z$), depolarizing ($\hat{\sigma}^x, \hat{\sigma}^y, \hat{\sigma}^z$).
- **Steady states**: Fixed points $\mathcal{L}(\rho_\infty) = 0$; often thermal equilibrium or ground states.
- **Quantum jump picture**: Lindblad equation averages over random quantum trajectories; jumps occur at Poisson-distributed times.
- **Quantum regression theorem**: Two-time correlations follow same master equation as populations.
