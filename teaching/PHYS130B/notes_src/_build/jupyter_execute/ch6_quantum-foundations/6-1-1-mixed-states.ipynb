## Lecture Notes

Not all quantum states are pure; sometimes we possess only incomplete information about a system, or it has interacted with an inaccessible environment. The density matrix elegantly captures this ignorance, unifying our description: whether we lack information by choice or by necessity, the formalism treats both identically. This single operator becomes the bridge between pure quantum coherence and classical uncertainty.

### Pure States via Density Matrix

For a **pure state** $|\psi\rangle$, the **density operator** is:

$$

\rho = |\psi\rangle\langle\psi|

$$ (eq-mixed-pure-density)

This is an **outer product** (rank-1 projection operator).

:::{admonition} Example: Qubit in Superposition
:class: example


Let $|\psi\rangle = \alpha |{\uparrow}\rangle + \beta |{\downarrow}\rangle$ with $|\alpha|^2 + |\beta|^2 = 1$.

$$

|\psi\rangle\langle\psi| = \begin{pmatrix} \alpha \\ \beta \end{pmatrix} \begin{pmatrix} \alpha^* & \beta^* \end{pmatrix} = \begin{pmatrix} |\alpha|^2 & \alpha^*\beta \\ \beta^*\alpha & |\beta|^2 \end{pmatrix}

$$ (eq-mixed-qubit-density)

**Interpretation**:
- **Diagonal elements** (populations): $|\alpha|^2$ = probability of measuring $|{\uparrow}\rangle$; $|\beta|^2$ = probability of measuring $|{\downarrow}\rangle$.
- **Off-diagonal elements** (coherences): $\alpha^*\beta, \beta^*\alpha$ = quantum interference terms; reflect superposition.

If $\alpha = 1/\sqrt{2}, \beta = 1/\sqrt{2}$ (superposition of $|{\uparrow}\rangle + |{\downarrow}\rangle$):

$$

\rho = \frac{1}{2}\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix}

$$ (eq-mixed-superposition-density)

Off-diagonal terms $\rho_{\uparrow\downarrow} = 1/2$ indicate coherence between states.

:::
### Properties of Pure-State Density Matrices

For a pure state $|\psi\rangle$, the density matrix $\rho = |\psi\rangle\langle\psi|$ satisfies fundamental properties:

1. **Hermiticity**: $\rho^\dagger = \rho$
2. **Trace normalization**: $\text{Tr}(\rho) = 1$
3. **Idempotence**: $\rho^2 = \rho$ (equivalently: $\text{Tr}(\rho^2) = 1$)
4. **Projection**: Eigenvalues are 0 and 1; eigenvector for eigenvalue 1 is $|\psi\rangle$

These properties are characteristic of **rank-1 projectors** and distinguish pure states from mixed states.

:::{admonition} Proof: Properties of Pure-State Density Matrices
:class: dropdown note

**Detailed derivations:**

1. **Hermiticity**: $\rho^\dagger = (|\psi\rangle\langle\psi|)^\dagger = |\psi\rangle\langle\psi| = \rho$. ✓

2. **Trace normalization**: $\text{Tr}(\rho) = \text{Tr}(|\psi\rangle\langle\psi|) = \langle\psi|\psi\rangle = 1$. ✓

3. **Idempotence**: $\rho^2 = (|\psi\rangle\langle\psi|)(|\psi\rangle\langle\psi|) = |\psi\rangle(\langle\psi|\psi\rangle)\langle\psi| = |\psi\rangle\langle\psi| = \rho$.
   Thus $\text{Tr}(\rho^2) = \text{Tr}(\rho) = 1$. ✓

4. **Projection**: Eigenvalues of $\rho$ are 0 and 1. The nonzero eigenvalue is 1 with eigenvector $|\psi\rangle$.

:::

### Mixed States and Ensembles

Suppose we **don't know** which pure state the system is in. Instead:
- With probability $p_1$, system is in state $|\psi_1\rangle$.
- With probability $p_2$, system is in state $|\psi_2\rangle$.
- ...
- With probability $p_n$, system is in state $|\psi_n\rangle$.

where $\sum_i p_i = 1$ and $p_i \geq 0$. This is a **statistical ensemble**.

The **density matrix** of the ensemble is:

$$

\rho = \sum_i p_i |\psi_i\rangle\langle\psi_i|

$$ (eq-mixed-ensemble-density)

If $|\psi_i\rangle$ are **orthonormal**, this is the **spectral decomposition** of $\rho$. More generally, the $|\psi_i\rangle$ need not be orthogonal (ensemble ambiguity, discussed below).

:::{admonition} Example: Maximally Mixed Qubit
:class: example


Ensemble: 50% chance of $|{\uparrow}\rangle$, 50% chance of $|{\downarrow}\rangle$.

$$

\rho = \frac{1}{2}|{\uparrow}\rangle\langle{\uparrow}| + \frac{1}{2}|{\downarrow}\rangle\langle{\downarrow}| = \frac{1}{2}\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = \frac{I}{2}

$$ (eq-mixed-maxmix-qubit)

**No coherences**: Off-diagonal elements are zero. The reduced density matrix is **diagonal** (classical-like).

:::
:::{admonition} Example: Thermal State
:class: example


System at temperature $T$ in equilibrium with a heat bath. Probability of energy eigenstate $|n\rangle$ is:

$$

p_n = \frac{e^{-E_n / k_B T}}{Z}, \quad Z = \sum_m e^{-E_m / k_B T} \quad \text{(partition function)}

$$ (eq-mixed-thermal-prob)

Density matrix (in energy eigenbasis):

$$

\rho_\text{thermal} = \sum_n \frac{e^{-E_n / k_B T}}{Z} |n\rangle\langle n| = \frac{e^{-H/k_B T}}{Z}

$$ (eq-mixed-thermal-density)

This is **diagonal in the energy basis** (no coherence at thermal equilibrium). Diagonal elements decrease with energy.

:::
### Properties of Valid Density Matrices

Any valid quantum state (pure or mixed) is represented by a density matrix $\rho$ satisfying:

**1. Hermiticity:**

$$

\rho^\dagger = \rho \quad \Rightarrow \quad \text{eigenvalues} \in \mathbb{R}

$$ (eq-mixed-hermitian)

**2. Trace normalization:**

$$

\text{Tr}(\rho) = 1 \quad \Rightarrow \quad \sum_i \lambda_i = 1

$$ (eq-mixed-trace-norm)

where $\lambda_i$ are eigenvalues. This ensures probability conservation.
:::


**3. Positive semidefiniteness:**

$$

\rho \geq 0 \quad \Rightarrow \quad \lambda_i \geq 0 \text{ for all } i

$$ (eq-mixed-positive)

Equivalently: $\langle\phi|\rho|\phi\rangle \geq 0$ for all $|\phi\rangle$. This ensures probabilities are non-negative.

### Purity and the Trace of the Square

The **purity** of a density matrix is defined as:

$$

\mathcal{P} = \text{Tr}(\rho^2)

$$ (eq-mixed-purity)

**Physical interpretation**: In the eigenbasis of $\rho$ (with eigenvalues $\lambda_i$):

$$

\text{Tr}(\rho^2) = \sum_i \lambda_i^2

$$ (eq-mixed-purity-eigenbasis)

#### Pure vs. Mixed

- **Pure state** ($|\psi\rangle\langle\psi|$): One eigenvalue = 1, rest = 0. Thus $\text{Tr}(\rho^2) = 1$.
- **Maximally mixed** (dimension $d$): All eigenvalues = $1/d$. Thus $\text{Tr}(\rho^2) = d \cdot (1/d)^2 = 1/d$.
- **General**: $1/d \leq \text{Tr}(\rho^2) \leq 1$.

**Interpretation**: 
- $\text{Tr}(\rho^2)$ close to 1: State is nearly pure; little uncertainty.
- $\text{Tr}(\rho^2)$ close to $1/d$: State is nearly maximally mixed; maximum uncertainty.

#### Alternative: Purity as Reciprocal

Sometimes defined as:

$$

\gamma = \text{Tr}(\rho^2) \quad \text{or} \quad \text{Purity} = d \cdot \text{Tr}(\rho^2)

$$

to normalize to $[0, 1]$ or $[1/d, 1]$ depending on convention.

### Ensemble Ambiguity

**Surprising fact**: Different ensembles can produce the **same** density matrix.

:::{admonition} Example: #### Example
:class: example


**Ensemble 1**: 
- 100% of state $|{\uparrow}\rangle \langle{\uparrow}|$ (pure, definitely spin-up).
- $\rho_1 = |{\uparrow}\rangle\langle{\uparrow}| = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}$.

**Ensemble 2**:
- 50% of state $|\rightarrow\rangle = \frac{1}{\sqrt{2}}(|{\uparrow}\rangle + |{\downarrow}\rangle)$.
- 50% of state $|\leftarrow\rangle = \frac{1}{\sqrt{2}}(|{\uparrow}\rangle - |{\downarrow}\rangle)$.

Compute $\rho_2$:

$$

\rho_2 = \frac{1}{2}|\rightarrow\rangle\langle\rightarrow| + \frac{1}{2}|\leftarrow\rangle\langle\leftarrow|

$$

$$

= \frac{1}{2} \cdot \frac{1}{2}\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix} + \frac{1}{2} \cdot \frac{1}{2}\begin{pmatrix} 1 & -1 \\ -1 & 1 \end{pmatrix}

$$

$$

= \frac{1}{4}\begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix} = \begin{pmatrix} 1/2 & 0 \\ 0 & 1/2 \end{pmatrix}

$$

Wait, this gives $\rho_2 = I/2$ (maximally mixed), **not** $\rho_1$. Let me reconsider.

**Better example**: Ensemble 1 and Ensemble 2 give different physics, so they won't both yield $\rho_1$.

Instead, consider:

**Ensemble A**: 50% spin-up along $z$; 50% spin-down along $z$:

$$

\rho_A = \frac{1}{2}|{\uparrow}\rangle\langle{\uparrow}| + \frac{1}{2}|{\downarrow}\rangle\langle{\downarrow}| = \frac{I}{2}

$$

**Ensemble B**: 50% spin-right along $x$; 50% spin-left along $x$:

$$

\rho_B = \frac{1}{2}|\rightarrow\rangle\langle\rightarrow| + \frac{1}{2}|\leftarrow\rangle\langle\leftarrow| = \frac{I}{2}

$$

**Both give $\rho = I/2$** (maximally mixed). But physical ensembles differ: in A, outcomes along $z$ are 50/50 deterministic (given ensemble); in B, outcomes along $x$ are 50/50.

**Key insight**: The density matrix alone **cannot distinguish** between these ensembles. Physical predictions (for observables in any basis) are identical; only preparation history differs.

:::
#### Spectral Decomposition

If $\rho = \sum_i \lambda_i |u_i\rangle\langle u_i|$ where $\{|u_i\rangle\}$ are eigenvectors and $\lambda_i \geq 0$ are eigenvalues, this is the **unique** spectral decomposition. All other decompositions $\rho = \sum_j p_j |\psi_j\rangle\langle\psi_j|$ (with non-orthogonal $|\psi_j\rangle$) represent different ensembles yielding the same $\rho$.

### The Bloch Ball for Qubit Mixed States

A qubit mixed state can be written:

$$

\rho = \frac{1}{2}(I + \boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}})

$$ (eq-mixed-bloch-density)

where $\boldsymbol{r} = (r_x, r_y, r_z)$ is the **Bloch vector** and $\boldsymbol{\hat{\sigma}} = (\hat{\sigma}_x, \hat{\sigma}_y, \hat{\sigma}_z)$ are Pauli matrices.

**Bloch sphere constraints**:

$$

|\boldsymbol{r}| \leq 1

$$ (eq-mixed-bloch-constraint)

- **Pure states**: $|\boldsymbol{r}| = 1$ (surface of Bloch sphere).
- **Mixed states**: $|\boldsymbol{r}| < 1$ (interior of Bloch ball).
- **Maximally mixed**: $\boldsymbol{r} = \boldsymbol{0}$ (center of ball); $\rho = I/2$.

#### Purity in Terms of Bloch Vector

The purity of a density matrix in terms of the Bloch vector is:

$$

\text{Tr}(\rho^2) = \frac{1}{2}(1 + |\boldsymbol{r}|^2)

$$ (eq-mixed-purity-bloch)

This formula connects the purity to the magnitude of the Bloch vector:

- **Pure state**: $|\boldsymbol{r}| = 1 \Rightarrow \text{Tr}(\rho^2) = 1$
- **Maximally mixed**: $|\boldsymbol{r}| = 0 \Rightarrow \text{Tr}(\rho^2) = 1/2$

The Bloch vector norm directly measures how "pure" the quantum state is. A state at the surface of the Bloch sphere ($|\boldsymbol{r}| = 1$) is pure; deeper in the interior, it becomes increasingly mixed.

:::{admonition} Derivation: Purity in Terms of Bloch Vector
:class: dropdown note

**Detailed calculation:**

Starting from $\rho = \frac{1}{2}(I + \boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}})$:

$$

\rho^2 = \frac{1}{4}(I + \boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}})^2 = \frac{1}{4}(I + 2\boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}} + (\boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}})^2)

$$

Using $(\boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}})^2 = |\boldsymbol{r}|^2 I$ (property of Pauli matrices):

$$

\rho^2 = \frac{1}{4}(I + 2\boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}} + |\boldsymbol{r}|^2 I) = \frac{1}{4}((1 + |\boldsymbol{r}|^2) I + 2\boldsymbol{r} \cdot \boldsymbol{\hat{\sigma}})

$$

Taking the trace:

$$

\text{Tr}(\rho^2) = \frac{1}{4}(2(1 + |\boldsymbol{r}|^2)) = \frac{1}{2}(1 + |\boldsymbol{r}|^2)

$$

:::{admonition} Example: Mixed Qubit State
:class: example

Consider $\rho = 0.8|{\uparrow}\rangle\langle{\uparrow}| + 0.2|{\downarrow}\rangle\langle{\downarrow}|$ (mixture favoring spin-up).

Eigenvalues: $\lambda_1 = 0.8, \lambda_2 = 0.2$.

$$

\boldsymbol{r} = (0, 0, \lambda_1 - \lambda_2) = (0, 0, 0.6)

$$

Thus $\rho = \frac{1}{2}(I + 0.6 \hat{\sigma}_z) = \begin{pmatrix} 0.8 & 0 \\ 0 & 0.2 \end{pmatrix}$. ✓

Bloch vector magnitude: $|\boldsymbol{r}| = 0.6 < 1$ (inside Bloch ball; mixed state). ✓

:::

### Summary

- **Pure state**: $\rho = |\psi\rangle\langle\psi|$; $\text{Tr}(\rho^2) = 1$; single rank-1 projection.
- **Mixed state**: $\rho = \sum_i p_i |\psi_i\rangle\langle\psi_i|$ (ensemble); $\text{Tr}(\rho^2) < 1$; rank $> 1$ generally.
- **Density operator properties**: Hermitian, $\text{Tr}(\rho) = 1$, positive semidefinite.
- **Purity** $\mathcal{P} = \text{Tr}(\rho^2)$: Measures mixedness; $\in [1/d, 1]$ (1 = pure, $1/d$ = maximally mixed).
- **Ensemble ambiguity**: Multiple ensembles yield same $\rho$; density matrix is all that matters for predictions.
