## Lecture Notes

For decades, physicists debated whether quantum mechanics was truly the final word, or whether hidden variables lurking beneath the surface determined outcomes in advance. Bell's insight was to ask a deceptively simple question: can any classical theory, however clever, reproduce the statistical correlations that quantum mechanics predicts? The answer, confirmed by countless experiments, is a resounding no—and it fundamentally transformed our understanding of reality itself.

### The EPR Paradox (1935)



:::{admonition} Discussion
:class: dropdown tip

The CHSH inequality gives max classical value 2 and quantum maximum 2√2. But there exist "super-quantum" correlations (PR boxes) that achieve value 4. Why doesn't nature allow PR boxes?
:::


Einstein, Podolsky, and Rosen proposed a thought experiment:

**Setup**: Entangled pair of particles; Alice measures particle 1, Bob measures particle 2 at a distant location.

**Observation**: Measurement on particle 1 affects the statistics for particle 2 (apparent correlation), seemingly instantaneously.

- Perform **alternative measurements** on particle 1 (spin along $z$ or $x$ axis) and observe perfect anticorrelations in the singlet state.
- Since particles are far apart, particle 2 cannot "know" which measurement Alice made.
- Therefore, the outcome for particle 2 must be **predetermined** (hidden variable).
- **Conclusion**: Quantum mechanics is incomplete; hidden variables determine outcomes, not wave function collapse.

**Quantum response**: Wave function incompleteness doesn't necessarily mean hidden variables; quantum correlations are real physical features.

### Local Hidden Variables

A **hidden variable theory** assumes each particle carries information $\lambda$ (hidden state) determining measurement outcomes deterministically or stochastically.

- **Locality**: Measurement outcome at location A depends only on $\lambda_A$ (and local measurement settings), not on distant measurement settings at B.
- **Realism**: Physical properties exist independent of observation; $\lambda$ encodes all properties.
- **No faster-than-light signals**: Information cannot propagate faster than light.

**Mathematical form**: For a measurement of observable $O_a$ on particle A and $O_b$ on particle B:

$$


\text{Outcome}_A = f_A(a, \lambda), \quad \text{Outcome}_B = f_B(b, \lambda)

$$ (eq-bell-local-hidden-vars)

where $a, b$ are measurement settings (angles), $\lambda$ is hidden variable, and functions $f$ depend only on local information.

### Bell Inequalities: CHSH Form

**Clauser–Horne–Shimony–Holt (CHSH) inequality** is the most commonly tested form of Bell inequalities:

$$


\left| \langle A_1 B_1 \rangle + \langle A_1 B_2 \rangle + \langle A_2 B_1 \rangle - \langle A_2 B_2 \rangle \right| \leq 2

$$ (eq-bell-chsh-inequality)

where:
- $A_i$ = outcome of Alice's measurement along direction $a_i$ (values $\pm 1$).
- $B_j$ = outcome of Bob's measurement along direction $b_j$ (values $\pm 1$).
- $\langle A_i B_j \rangle$ = expectation value of product (measured over many runs with $\lambda$ distributed according to hidden variable distribution $\rho(\lambda)$).

To derive the CHSH inequality:

:::{admonition} Derivation: CHSH Inequality
:class: dropdown information

Assume local hidden variables. For any hidden variable $\lambda$ and measurement settings:

$$


A_1(\lambda) B_1(\lambda) + A_1(\lambda) B_2(\lambda) + A_2(\lambda) B_1(\lambda) - A_2(\lambda) B_2(\lambda)

$$


Rewrite as:

$$


A_1(\lambda)[B_1(\lambda) + B_2(\lambda)] + A_2(\lambda)[B_1(\lambda) - B_2(\lambda)]

$$


Since $A_1, A_2 \in \{-1, +1\}$ and $B_1, B_2 \in \{-1, +1\}$:
- If $B_1(\lambda) + B_2(\lambda) = \pm 2$, first term contributes $\pm 2$; second is at most 2 in magnitude.
- If $B_1(\lambda) - B_2(\lambda) = \pm 2$, second term contributes $\pm 2$; first is at most 2 in magnitude.
- Otherwise, both terms contribute at most 1 each in the right regime.

Overall: The expression is bounded by 2. Averaging over $\lambda$: $\langle | \cdots | \rangle \leq 2$.
:::
### Quantum Prediction for Singlet State

Consider the singlet state:

$$


|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|{01}\rangle - |{10}\rangle)

$$ (eq-bell-singlet)

**Measurement correlations**: If Alice measures spin along direction $\boldsymbol{a}$ and Bob along $\boldsymbol{b}$, the correlation is:

$$


\langle A(\boldsymbol{a}) B(\boldsymbol{b}) \rangle = -\boldsymbol{a} \cdot \boldsymbol{b}

$$ (eq-bell-singlet-correlation)

**Proof**: The singlet is rotationally invariant and antisymmetric. For parallel measurements ($\boldsymbol{a} = \boldsymbol{b}$): outcomes are always opposite, giving $\langle AB \rangle = -1$. For perpendicular ($\boldsymbol{a} \perp \boldsymbol{b}$): outcomes are uncorrelated on average, giving $\langle AB \rangle = 0$. The general formula is $-\cos(\theta)$ where $\theta$ is the angle between $\boldsymbol{a}$ and $\boldsymbol{b}$; in 2D (equatorial plane), $-\boldsymbol{a} \cdot \boldsymbol{b}$.

#### Optimal Choice of Angles

Choose measurement angles to maximize the CHSH expression:
- $a_1 = 0°$ (Alice's first axis).
- $a_2 = 45°$ (Alice's second axis).
- $b_1 = 22.5°$ (Bob's first axis).
- $b_2 = 67.5°$ (Bob's second axis).

Compute correlations using $\langle AB \rangle = -\cos(\Delta\theta)$:

$$


\langle A_1 B_1 \rangle = -\cos(22.5°) = -\cos(\pi/8) \approx -0.924

$$


$$

\langle A_1 B_2 \rangle = -\cos(67.5°) = -\cos(3\pi/8) \approx -0.383

$$


$$

\langle A_2 B_1 \rangle = -\cos(22.5°) \approx -0.924

$$


$$

\langle A_2 B_2 \rangle = -\cos(67.5°) \approx -0.383

$$ (eq-bell-optimal-correlations)

**CHSH value**:

$$


| -0.924 - 0.383 - 0.924 + 0.383 | = | -1.848 | \approx 1.848

$$


**Better approach using $\cos(22.5°) = \sqrt{(1+\cos(45°))/2} = \sqrt{(2+\sqrt{2})/4}$**:

$$


\text{CHSH} = 2\sqrt{2} \approx 2.828 > 2

$$ (eq-bell-quantum-violation)

**Interpretation**: Quantum prediction violates CHSH inequality by a factor $2\sqrt{2}/2 = \sqrt{2} \approx 1.414$. Violation is **robust** and independent of angle choices (within symmetry-respecting parametrizations).

:::{admonition} Tsirelson Bound: Maximum Quantum Violation
:class: important

The maximum value of the CHSH correlator achievable by quantum mechanics is:

$$


\text{CHSH}_\text{max}^\text{quantum} = 2\sqrt{2} \approx 2.828

$$


**Classical bound**: $\leq 2$ (local hidden variables)  
**Quantum bound (Tsirelson)**: $\leq 2\sqrt{2}$ (any quantum state)  
**Hypothetical PR-box**: $4$ (super-quantum, forbidden by physics)

This quantum maximum is achieved by maximally entangled two-qubit states (Bell states) measured along optimal directions. The Tsirelson bound represents the ultimate limit of quantum nonlocal correlations.

:::

### Interpretation: Nonlocality

The CHSH violation rules out **local hidden variable theories**. Bell's theorem concludes:

:::{admonition} Bell's Theorem
:class: important

No local hidden variable theory can reproduce all predictions of quantum mechanics. Either:
1. Nature is **nonlocal** (information/correlations can propagate instantaneously), or
2. Hidden variables don't exist (quantum indeterminacy is fundamental).

:::

**Nonlocality in quantum mechanics**:
- Measurement outcomes are **correlated** over arbitrary distances instantaneously.
- Correlations are **real** (not due to pre-arrangement or classical communication).
- **No signaling possible** (see no-communication theorem below).
- Cannot be reproduced by any theory respecting local causality.

### Bell States

The four Bell states are maximally entangled and all violate Bell inequalities with the same maximum violation $2\sqrt{2}$:

$$


|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|{00}\rangle + |{11}\rangle) \quad \text{(triplet, parallel)}

$$ (eq-bell-bell-states-1)

$$


|\Phi^-\rangle = \frac{1}{\sqrt{2}}(|{00}\rangle - |{11}\rangle)

$$ (eq-bell-bell-states-2)

$$


|\Psi^+\rangle = \frac{1}{\sqrt{2}}(|{01}\rangle + |{10}\rangle)

$$ (eq-bell-bell-states-3)

$$


|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|{01}\rangle - |{10}\rangle) \quad \text{(singlet, antiparallel)}

$$ (eq-bell-bell-states-4)

All four form an **orthonormal basis** for two qubits (complete basis for Bell measurements).

### Loophole-Free Bell Tests

**Historical loopholes** in Bell tests:

- **Detection loophole**: Not all photons detected; undetected particles could have predetermined outcomes consistent with local theory.
- **Locality loophole**: Measurement choices not sufficiently spacelike separated or randomized; could be correlated with hidden variables.
- **Freedom-of-choice loophole**: Measurement settings could be predetermined by hidden variables (superdeterminism).

**Loophole-free tests (2015+)**:
- **2015 BIG Bell Test** (multiple labs): Closed detection and locality loopholes.
- **2015 Zhang et al. (Science)**: Closed locality loophole with long-distance entangled atoms.
- **2022 experiments (Nature)**: High-efficiency photonic sources and detectors; all loopholes closed simultaneously.

**Consensus**: Quantum mechanics is correct; local hidden variables ruled out with very high confidence ($> 5\sigma$ in recent tests).

### No-Communication Theorem

**Paradox**: Bell inequalities show nonlocal correlations; why can't Alice signal to Bob faster than light?

**Answer**: The **no-communication theorem** states:

:::{admonition} No-Communication Theorem
:class: important

No information can be transmitted faster than light using entanglement and local operations, even though measurement outcomes are correlated.

:::

**Proof sketch**: Alice measures her qubit in the singlet state $|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|{01}\rangle - |{10}\rangle)$ and obtains outcome $\pm 1$ at random. Bob's reduced density matrix **before** he learns Alice's result is:

$$


\rho_B = \frac{I}{2} \quad \text{(maximally mixed)}

$$ (eq-bell-no-comm-reduced)

This is **independent** of Alice's measurement basis (up to local unitary). Bob measures any observable on his qubit:

$$


\langle O_B \rangle_B = \text{Tr}(O_B \rho_B) \quad \text{(independent of Alice's choice)}

$$ (eq-bell-no-comm-expectation)

Bob's local statistics are **unchanged** regardless of what Alice does. The correlation only becomes apparent when Alice and Bob **compare results** via classical communication (which is limited by light speed).

**Metaphor**: Alice sees random heads/tails; Bob sees random heads/tails. Only when they meet and compare do they realize their outcomes are perfectly correlated. The correlation doesn't transmit information; it's established by the pre-shared entangled state.

### Quantum Teleportation and Bell Measurement

Bell measurement (projecting onto Bell basis) is essential for quantum teleportation:

- **Bell measurement** by Alice on qubits (data + entangled partner): 2 classical bits of information.
- **Send 2 bits** classically to Bob.
- **Bob applies correction** (Pauli gate) to his half of entangled pair.

Result: Unknown state teleported to Bob, with no faster-than-light signaling (requires classical bits).

### Contextuality (Brief Mention)

**Kochen–Specker theorem** (1967): In quantum mechanics, **measurement outcomes cannot be assigned definite values** independent of the context (choice of measurement basis). This is a weaker form of nonlocality that holds even for single systems.

**Implication**: Quantum indeterminacy is intrinsic, not due to hidden variables or measurement disturbance alone.

### Summary

- **EPR paradox**: Einstein questioned completeness of quantum mechanics; suggested hidden variables.
- **Local hidden variables**: Assume outcomes predetermined; measurement settings cannot influence distant outcomes.
- **Bell inequalities** (CHSH): $|\langle A_1 B_1 \rangle + \langle A_1 B_2 \rangle + \langle A_2 B_1 \rangle - \langle A_2 B_2 \rangle| \leq 2$ in local theories.
- **Quantum violation**: Singlet and Bell states achieve CHSH $= 2\sqrt{2} \approx 2.828 > 2$.
- **Bell's theorem**: No local hidden variable theory can reproduce quantum predictions. Nature is **nonlocal**.
- **Loophole-free tests**: Modern experiments (2015+) confirm quantum mechanics; local theories ruled out.
- **No-communication**: Nonlocal correlations don't permit faster-than-light signaling. Comparison requires classical communication.
- **Bell basis**: Four orthogonal states used for Bell measurements; complete basis for two qubits.
- **Contextuality**: Measurement outcomes context-dependent; quantum indeterminacy is fundamental (Kochen–Specker).
