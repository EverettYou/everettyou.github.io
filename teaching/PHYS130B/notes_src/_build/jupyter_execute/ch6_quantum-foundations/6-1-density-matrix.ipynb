# 6.1 Density Matrix


### Overview

Until now, we have described quantum states as **vectors** $|\psi\rangle$ in Hilbert space. This works perfectly for **pure states**—states with complete, maximal information.

However, quantum mechanics also deals with **mixed states**:

- States arising from **incomplete knowledge** (we don't know exactly what state the system is in).
- States arising from **entanglement** with an inaccessible environment (the system is part of a larger system, but we only observe a subsystem).
- States resulting from **decoherence** due to uncontrolled interactions with the environment.

For these cases, a single state vector is insufficient. We introduce the **density matrix** (or **density operator**).

The **density matrix** is a Hermitian operator $\rho$ satisfying:

1. **Hermiticity**: $\rho^\dagger = \rho$
2. **Positivity**: All eigenvalues are non-negative ($\rho \geq 0$)
3. **Trace normalization**: $\text{Tr}(\rho) = 1$

For a **pure state** $|\psi\rangle$:

$$

\rho = |\psi\rangle\langle\psi|

$$ (eq-6-1-pure-density)

For a **mixed state** (probabilistic ensemble), if the system is in state $|\psi_i\rangle$ with probability $p_i$:

$$

\rho = \sum_i p_i |\psi_i\rangle\langle\psi_i|

$$ (eq-6-1-mixed-density)

The **purity** $\mathcal{P} = \text{Tr}(\rho^2)$ distinguishes pure ($\mathcal{P}=1$) from mixed ($\mathcal{P}<1$) states.

For an observable $\hat{O}$, the expectation value in state $\rho$ is:

$$

\langle \hat{O} \rangle = \text{Tr}(\rho \hat{O})

$$ (eq-6-1-expectation)

This formula unifies pure and mixed states: for $\rho = |\psi\rangle\langle\psi|$, it reduces to $\langle\psi|\hat{O}|\psi\rangle$.
