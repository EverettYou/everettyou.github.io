# 6.2 Entanglement


### Overview

Entanglement is the defining feature of quantum mechanics that has no classical counterpart. When two quantum systems become entangled, the state of the whole cannot be decomposed into independent states of the parts. This leads to correlations that violate classical bounds, as demonstrated by Bell's theorem, and forms the basis of quantum information science.
