## Lecture Notes



The Heisenberg uncertainty principle is not about the limitations of our instruments; it is a deep feature of quantum reality stemming from the non-commutativity of observables. When two observables don't commute, we cannot know both their values simultaneously—gaining knowledge of one necessarily destroys knowledge of the other.

### The Commutator

For two operators $\hat{A}$ and $\hat{B}$, the **commutator** is defined as

$$


[\hat{A}, \hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}

$$ (eq-commutator-def)

:::{admonition} Commutator Definition
:class: important

The **commutator** $[\hat{A}, \hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}$ measures whether two operators commute:
- If $[\hat{A}, \hat{B}] = 0$, the operators **commute**: they can be applied in any order.
- If $[\hat{A}, \hat{B}] \neq 0$, the operators **don't commute**: the order matters.

This is the fundamental object that determines whether two observables can be measured simultaneously.

:::

**Interpretation**:
- If $[\hat{A}, \hat{B}] = 0$ ("they commute"), then the order of applying the operators doesn't matter: $\hat{A}\hat{B} = \hat{B}\hat{A}$.
- If $[\hat{A}, \hat{B}] \neq 0$ ("they don't commute"), the order matters: $\hat{A}\hat{B} \neq \hat{B}\hat{A}$.

#### Properties of Commutators

1. **Antisymmetry**: $[\hat{A}, \hat{B}] = -[\hat{B}, \hat{A}]$
2. **Bilinearity**: $[c_1\hat{A}_1 + c_2\hat{A}_2, \hat{B}] = c_1[\hat{A}_1, \hat{B}] + c_2[\hat{A}_2, \hat{B}]$
3. **Jacobi identity**: $[\hat{A}, [\hat{B}, \hat{C}]] + [\hat{B}, [\hat{C}, \hat{A}]] + [\hat{C}, [\hat{A}, \hat{B}]] = 0$
4. **Self-commutation**: $[\hat{A}, \hat{A}] = 0$

#### Commutators of Pauli Matrices

The commutation relations among Pauli matrices are fundamental:

$$


[\hat{\sigma}^i, \hat{\sigma}^j] = 2\mathrm{i}\epsilon_{ijk}\hat{\sigma}^k

$$ (eq-pauli-commutators)

where $\epsilon_{ijk}$ is the **Levi-Civita symbol** (totally antisymmetric in all indices).

**Examples**:

$$


[\hat{\sigma}^x, \hat{\sigma}^y] = 2\mathrm{i}\hat{\sigma}^z

$$


:::{admonition} Verification: Commutation Relation
:class: dropdown note

$$


\hat{\sigma}^x \hat{\sigma}^y = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 0 & -\mathrm{i} \\ i & 0 \end{pmatrix} = \begin{pmatrix} i & 0 \\ 0 & -i \end{pmatrix}

$$


$$

\hat{\sigma}^y \hat{\sigma}^x = \begin{pmatrix} 0 & -\mathrm{i} \\ i & 0 \end{pmatrix} \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} = \begin{pmatrix} -i & 0 \\ 0 & i \end{pmatrix}

$$


$$

[\hat{\sigma}^x, \hat{\sigma}^y] = \begin{pmatrix} i & 0 \\ 0 & -i \end{pmatrix} - \begin{pmatrix} -i & 0 \\ 0 & i \end{pmatrix} = \begin{pmatrix} 2i & 0 \\ 0 & -2i \end{pmatrix} = 2i \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} = 2\mathrm{i}\hat{\sigma}^z \quad \checkmark

$$


:::

**All commutation relations**:

$$


[\hat{\sigma}^x, \hat{\sigma}^y] = 2\mathrm{i}\hat{\sigma}^z, \quad [\hat{\sigma}^y, \hat{\sigma}^z] = 2\mathrm{i}\hat{\sigma}^x, \quad [\hat{\sigma}^z, \hat{\sigma}^x] = 2\mathrm{i}\hat{\sigma}^y

$$


By antisymmetry, $[\hat{\sigma}^i, \hat{\sigma}^i] = 0$ for any $i$.

#### Anti-commutators

The **anti-commutator** is

$$


\{\hat{A}, \hat{B}\} = \hat{A}\hat{B} + \hat{B}\hat{A}

$$


**For Pauli matrices**:

$$


\{\hat{\sigma}^i, \hat{\sigma}^j\} = 2\delta^{ij} I

$$


where $\delta^{ij}$ is the Kronecker delta. This shows that Pauli matrices **anti-commute** when $i \neq j$.

:::{admonition} Theorem: Commuting Operators Share
:class: important

Two Hermitian operators $\hat{A}$ and $\hat{B}$ share the **same set of eigenvectors** (a common eigenbasis) if and only if $[\hat{A}, \hat{B}] = 0$.

**Consequence**: Non-commuting observables cannot have simultaneous eigenstates—no quantum state can be a simultaneous eigenstate of both.

:::

:::{admonition} Proof: Common Eigenbasis Implies
:class: dropdown note

Suppose $\hat{A}$ and $\hat{B}$ have the same eigenbasis $\{|\phi_i\rangle\}$ with eigenvalues $\lambda_i$ and $\mu_i$:

$$


\hat{A} |\phi_i\rangle = \lambda_i |\phi_i\rangle, \quad \hat{B} |\phi_i\rangle = \mu_i |\phi_i\rangle

$$


Then:

$$


[\hat{A}, \hat{B}] |\phi_i\rangle = (\hat{A}\hat{B} - \hat{B}\hat{A})|\phi_i\rangle = \lambda_i \mu_i |\phi_i\rangle - \mu_i \lambda_i |\phi_i\rangle = 0

$$


Since this holds for all basis vectors, $[\hat{A}, \hat{B}] = 0$.
:::

**Corollary**: Two non-commuting operators **cannot** share a complete set of eigenvectors. Therefore, a quantum state cannot be a simultaneous eigenstate of both operators.

:::{admonition} Example: $\hat{\sigma}^x$ and $\hat{\sigma}^z$
:class: example

The eigenbases of $\hat{\sigma}^x$ and $\hat{\sigma}^z$ are different:

- $\hat{\sigma}^z$ eigenbasis: $\{|0\rangle, |1\rangle\}$
- $\hat{\sigma}^x$ eigenbasis: $\{|+\rangle, |-\rangle\}$

Since $[\hat{\sigma}^x, \hat{\sigma}^z] = -2\mathrm{i}\hat{\sigma}^y \neq 0$, they don't commute. No state is a simultaneous eigenstate of both.

For example, $|0\rangle$ is an eigenstate of $\hat{\sigma}^z$ (eigenvalue +1) but **not** of $\hat{\sigma}^x$:

$$


\hat{\sigma}^x |0\rangle = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} 0 \\ 1 \end{pmatrix} = |1\rangle \neq \lambda |0\rangle

$$


The state is a superposition in the $\hat{\sigma}^x$ basis:

$$


|0\rangle = \frac{1}{\sqrt{2}}(|+\rangle + |-\rangle)

$$


:::

:::

### The Uncertainty Principle

The **Heisenberg uncertainty principle** relates the uncertainties (standard deviations) in measurements of two non-commuting observables.

:::{admonition} Heisenberg Uncertainty Principle
:class: important

**Theorem (Uncertainty Relation)**:

$$


\Delta \hat{A} \cdot \Delta \hat{B} \geq \frac{1}{2} |\langle [\hat{A}, \hat{B}] \rangle |

$$


This fundamental inequality states: **if two observables don't commute, they cannot be simultaneously measured to arbitrary precision.** The uncertainty is ontological—a property of nature itself, not epistemic—a limitation of measurement apparatus.

:::

$$ (eq-uncertainty-principle)

where:
- $\Delta \hat{A} = \sqrt{\langle \hat{A}^2 \rangle - \langle \hat{A} \rangle^2}$ is the standard deviation of $\hat{A}$
- $\Delta \hat{B} = \sqrt{\langle \hat{B}^2 \rangle - \langle \hat{B} \rangle^2}$ is the standard deviation of $\hat{B}$
- $\langle [\hat{A}, \hat{B}] \rangle = \langle \psi | [\hat{A}, \hat{B}] | \psi \rangle$ is the expectation value of the commutator

**Physical Interpretation**: If two observables don't commute (i.e., $[\hat{A}, \hat{B}] \neq 0$), then we cannot simultaneously have small uncertainties in both measurements. This is not a limitation of our experimental technique—it is a fundamental property of quantum systems.

For $\hat{\sigma}^x$ and $\hat{\sigma}^z$:

$$


[\hat{\sigma}^x, \hat{\sigma}^z] = -2\mathrm{i}\hat{\sigma}^y

$$


The uncertainty relation becomes:

$$


\Delta \hat{\sigma}^x \cdot \Delta \hat{\sigma}^z \geq \frac{1}{2}|\langle -2\mathrm{i}\hat{\sigma}^y \rangle| = |\langle \hat{\sigma}^y \rangle|

$$ (eq-uncertainty-pauli)

:::{admonition} Example: State with Definite X-spin
:class: dropdown note

For the state $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$:

**Z-component**:
- $\langle \hat{\sigma}^z \rangle = 0$
- $\langle (\hat{\sigma}^z)^2 \rangle = 1$, so $\Delta \hat{\sigma}^z = 1$

**X-component**:
- $\langle \hat{\sigma}^x \rangle = 1$ (state is eigenstate of $\hat{\sigma}^x$)
- $\Delta \hat{\sigma}^x = 0$

This state has **definite** x-spin but **maximum** uncertainty in z-spin. The product is $1 \cdot 0 = 0$, saturating the bound (since $\langle \hat{\sigma}^y \rangle = 0$).
:::


:::{admonition} Discussion
:class: dropdown warning

The Robertson uncertainty relation states $\Delta \hat{A} \cdot \Delta \hat{B} \geq \frac{1}{2}|\langle [\hat{A}, \hat{B}] \rangle|$ for any state.

### Complementarity and Sequential Measurements

The non-commutativity of observables has a striking experimental consequence.

**Gedankenexperiment**: Consider a qubit in state $|0\rangle$ (spin-up along z).

**Scenario A (measure only $\hat{\sigma}^z$)**:
- Measure $\hat{\sigma}^z$: get outcome +1 with probability 1.
- Result: 100% of atoms are spin-up along z.

**Scenario B (measure $\hat{\sigma}^x$ first, then $\hat{\sigma}^z$)**:
- Measure $\hat{\sigma}^x$: get outcome $+1$ with probability 1/2, outcome $-1$ with probability 1/2.
  - If outcome is +1, state collapses to $|+\rangle$.
  - If outcome is -1, state collapses to $|-\rangle$.
- Now measure $\hat{\sigma}^z$:
  - Starting from $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$: outcomes $\pm 1$ with equal probability (50-50).
  - Starting from $|-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$: outcomes $\pm 1$ with equal probability (50-50).
- Result: Overall, 25% of atoms are spin-up along z, 75% spin-down (or by symmetry, 50-50).

**Conclusion**: The intermediate $\hat{\sigma}^x$ measurement **destroys the z-information**. We cannot simultaneously be prepared with definite z-spin and then measure it with certainty after measuring x.

:::


:::{admonition} Discussion
:class: dropdown warning

**Sequential measurement and the no-cloning theorem.**

An ideal projective measurement of $\hat{\sigma}^z$ on $|{+x}\rangle = (|{+z}\rangle + |{-z}\rangle)/\sqrt{2}$ gives $+1$ or $-1$ with equal probability. After the measurement, the state collapses.

- If we wanted to "copy" the pre-measurement state $|{+x}\rangle$ before measuring, we could measure it and prepare a fresh $|{+x}\rangle$. Why does this NOT work for an unknown state? What is the no-cloning theorem?
- Suppose Alice has a particle in the state $\alpha|{+z}\rangle + \beta|{-z}\rangle$ and wants to tell Bob what $|\alpha|^2$ is (the probability of measuring $+z$). Why must she measure MANY identically prepared copies of the state to determine $|\alpha|^2$? What is the minimum number of measurements needed to determine the state to precision $\epsilon$?
- In the sequential measurement experiment: prepare $|{+x}\rangle$, measure $\hat{\sigma}^z$, obtain $|{+z}\rangle$, then measure $\hat{\sigma}^x$ again. What probability do you get for $\hat{\sigma}^x = +1$? Compare to measuring $\hat{\sigma}^x$ twice without intermediate $\hat{\sigma}^z$ measurement.
:::
### Observable Algebras and Quantum Complementarity

Two observables are **complementary** if their eigenbases are "maximally different." For a qubit:

- $\hat{\sigma}^z$ and $\hat{\sigma}^x$ are complementary: if a state is an eigenstate of one, measurements of the other give completely random results (50-50).
- This is because the eigenvectors form a **mutually unbiased basis** (MUB).

**Measurement of complementary observables**:
- Measuring $\hat{\sigma}^z$ gives a definite outcome: $\langle \psi | \hat{\sigma}^z | \psi \rangle = \pm 1$.
- Measuring $\hat{\sigma}^x$ on the same state gives random 50-50 outcomes.

This reflects the fundamental indeterminacy of quantum mechanics: not all properties can have simultaneous definite values.

### Summary

- **Commutators** quantify non-commutativity: $[\hat{A}, \hat{B}] = 0$ means the operators can be applied in any order.
- **Pauli matrices** have non-zero commutation relations $[\hat{\sigma}^i, \hat{\sigma}^j] = 2\mathrm{i}\epsilon_{ijk}\hat{\sigma}^k$, reflecting the incompatibility of spin components.
- **Non-commuting observables** cannot share an eigenbasis; no state is a simultaneous eigenstate of both.
- **Uncertainty principle**: $\Delta \hat{A} \cdot \Delta \hat{B} \geq \frac{1}{2}|\langle [\hat{A}, \hat{B}] \rangle|$ expresses fundamental incompatibility, not measurement imprecision.
- **Complementarity**: Measuring one observable destroys information about a complementary one. Sequential measurements of non-commuting observables yield different results than measuring only the first one.
