## Lecture Notes

So far, we have described quantum mechanics in the **Schrödinger picture**, where states evolve in time and observables remain fixed. This is intuitive for beginners, but there is a mathematically equivalent description—the **Heisenberg picture**—where operators evolve instead of states. This perspective is more natural when discussing conservation laws and conserved quantities, and it reveals deep connections between quantum and classical mechanics. In this section, we explore operator dynamics and meet the Heisenberg equation of motion.

### The Two Pictures: Schrödinger vs. Heisenberg

#### Schrödinger Picture

- States $|\psi(t)\rangle_S$ evolve according to the Schrödinger equation: $\mathrm{i}\hbar \frac{\mathrm{d}}{\mathrm{d}t}|\psi(t)\rangle_S = \hat{H}|\psi(t)\rangle_S$
- Observables $\hat{A}_S$ are time-independent
- Expectation values: $\langle A(t) \rangle = {}_S\langle\psi(t)|\hat{A}_S|\psi(t)\rangle_S$

#### Heisenberg Picture

- States $|\psi\rangle_H$ are fixed (we choose the initial state)
- Observables $\hat{A}_H(t)$ evolve in time
- Expectation values: $\langle A(t) \rangle = {}_H\langle\psi|\hat{A}_H(t)|\psi\rangle_H$

#:::{admonition} Equivalence of Pictures
:class: important

Both pictures give the **same expectation values** and observable predictions. They are related by a unitary transformation: if $U(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$, then

$$


\hat{A}_H(t) = U^\dagger(t) \hat{A}_S U(t) \quad \text{and} \quad |\psi\rangle_H = U^\dagger(t) |\psi(t)\rangle_S

$$


:::

#### When is Each Useful?

- **Schrödinger picture**: Easier for solving time-dependent problems with explicit states; natural for computing populations and transition probabilities.
- **Heisenberg picture**: Better for identifying conserved quantities and symmetries; more closely mirrors classical mechanics; essential for understanding the correspondence principle.

:::{admonition} Discussion  
:class: dropdown warning

In classical mechanics, Hamilton's equations relate the time derivative of a function on phase space to the Poisson bracket: $\frac{\mathrm{d}f}{\mathrm{d}t} = \{f, H\}$. 

The Heisenberg equation $\frac{\mathrm{d}\hat{O}_H}{\mathrm{d}t} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{O}_H]$ looks structurally identical, with the commutator replacing the Poisson bracket.

Does this mean quantum mechanics is "just" classical mechanics with a different algebra? Or does the commutator capture something fundamentally quantum that has no classical analogue? Can you think of a situation where the quantum and classical equations give the same time evolution? When do they differ qualitatively?

:::

#### Derivation

To find how operators evolve in the Heisenberg picture, we start with the relation $\hat{A}_H(t) = U^\dagger(t) \hat{A}_S U(t)$ and differentiate with respect to time:

$$


\frac{\mathrm{d}\hat{A}_H}{dt} = \frac{\mathrm{d} U^\dagger}{dt} \hat{A}_S U + U^\dagger \hat{A}_S \frac{\mathrm{d}U}{dt}

$$


Using $\frac{\mathrm{d}U}{dt} = -\frac{\mathrm{i}}{\hbar}\hat{H} U$ and $\frac{\mathrm{d}U^\dagger}{dt} = \frac{\mathrm{i}}{\hbar}U^\dagger \hat{H}$, we have

$$


\frac{\mathrm{d}\hat{A}_H}{dt} = \frac{\mathrm{i}}{\hbar}U^\dagger \hat{H} \hat{A}_S U - \frac{\mathrm{i}}{\hbar}U^\dagger \hat{A}_S \hat{H} U = \frac{\mathrm{i}}{\hbar}U^\dagger [\hat{H}, \hat{A}_S] U

$$


Since $\hat{H}$ commutes with both $U$ and $U^\dagger$ (it generates the evolution), this becomes the **Heisenberg equation of motion**.

#:::{admonition} Heisenberg: Equation
:class: important

$$


\frac{\mathrm{d}\hat{A}_H}{dt} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{A}_H] + \left(\frac{\partial \hat{A}}{\partial t}\right)_H

$$


The first term represents evolution due to commutation with the Hamiltonian. The second term accounts for explicit time-dependence in the operator itself; for time-independent observables, it vanishes.
:::

:::{admonition} Detailed Derivation
:class: dropdown note

Define the Heisenberg operator as $\hat{A}_H(t) = U^\dagger(t)\hat{A}_S U(t)$. Then:

$$


\frac{\mathrm{d}\hat{A}_H}{dt} = \left(\frac{\mathrm{d}U^\dagger}{dt}\right)\hat{A}_S U + U^\dagger\hat{A}_S\left(\frac{\mathrm{d}U}{dt}\right) + U^\dagger\left(\frac{\partial\hat{A}_S}{\partial t}\right)U

$$


From the Schrödinger equation, $\mathrm{i}\hbar\dot{U} = \hat{H}U$, so $\dot{U} = \frac{\mathrm{i}}{\hbar}\hat{H}U$ and $\dot{U}^\dagger = \frac{\mathrm{i}}{\hbar}U^\dagger\hat{H}$ (since $\hat{H}$ is Hermitian).

$$


\frac{\mathrm{d}\hat{A}_H}{dt} = \frac{\mathrm{i}}{\hbar}U^\dagger\hat{H}\hat{A}_S U - \frac{\mathrm{i}}{\hbar}U^\dagger\hat{A}_S\hat{H}U + U^\dagger\left(\frac{\partial\hat{A}_S}{\partial t}\right)U

$$


$$

= \frac{\mathrm{i}}{\hbar}U^\dagger[\hat{H}, \hat{A}_S]U + U^\dagger\left(\frac{\partial\hat{A}_S}{\partial t}\right)U = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{A}_H] + \left(\frac{\partial\hat{A}}{\partial t}\right)_H

$$


since $\hat{H}$ is time-independent and commutes with $U$ and $U^\dagger$.
:::

### Conserved Quantities and Conservation Laws

A key insight emerges from the Heisenberg equation: if an observable $\hat{A}$ has no explicit time-dependence and **commutes with the Hamiltonian**, then

$$


[\hat{A}, \hat{H}] = 0 \quad \Rightarrow \quad \frac{\mathrm{d}\hat{A}_H}{dt} = 0

$$


:::{admonition} Conserved Quantities
:class: important

An operator $\hat{A}$ is a **constant of motion** (conserved quantity) if and only if $[\hat{A}, \hat{H}] = 0$. In this case, its expectation value is time-independent:

$$


\frac{\mathrm{d}}{\mathrm{d}t}\langle\hat{A}\rangle = 0

$$


**In quantum mechanics, commuting with the Hamiltonian is the precise condition for a conserved quantity.** This is the quantum mechanical expression of Noether's theorem and Wigner's symmetry principle. For every continuous symmetry of the system, there is a conserved operator.
:::

#### Physical Examples

- If $[\hat{L}_z, \hat{H}] = 0$ (rotational symmetry about the z-axis), then angular momentum in the z-direction is conserved.
- If the Hamiltonian is time-independent (translational symmetry in time), then energy is conserved: $[\hat{H}, \hat{H}] = 0$ always.
- For a free particle, $[\hat{\boldsymbol{p}}, \hat{H}] = 0$, so momentum is conserved.

:::{admonition} Discussion
:class: dropdown warning

The Ehrenfest theorem says that the expectation values of position and momentum satisfy classical equations of motion:

$$


\frac{\mathrm{d}\langle\hat{x}\rangle}{\mathrm{d}t} = \frac{\langle\hat{p}\rangle}{m}, \quad \frac{\mathrm{d}\langle\hat{p}\rangle}{\mathrm{d}t} = -\langle\frac{\mathrm{d}V}{\mathrm{d}x}\rangle

$$


This seems to suggest that quantum systems obey classical laws on average. But notice the last term: it involves $\langle\frac{\mathrm{d}V}{\mathrm{d}x}\rangle$, not $\frac{\mathrm{d}V}{\mathrm{d}\langle x\rangle}$. These are equal only if $V$ is linear in $x$. 

For a particle in a nonlinear potential (like a quantum harmonic oscillator), the expectation values obey the classical equations *exactly*. For a particle encountering a barrier or potential well, the expectation values can exhibit genuinely quantum behavior. Why is the potential's shape crucial? Does Ehrenfest's theorem mean classical mechanics emerges from quantum mechanics, or is it just a formal similarity?

:::

#### Statement and Derivation

The Heisenberg equation connects quantum expectations to classical equations of motion. For a system with Hamiltonian $\hat{H} = \frac{\hat{p}^2}{2m} + V(\hat{x})$, applying the Heisenberg equation to position and momentum gives

$$


\frac{\mathrm{d}}{\mathrm{d}t}\langle\hat{x}\rangle = \frac{\mathrm{i}}{\hbar}\langle[\hat{H}, \hat{x}]\rangle = \frac{1}{m}\langle\hat{p}\rangle

$$


$$

\frac{\mathrm{d}}{\mathrm{d}t}\langle\hat{p}\rangle = \frac{\mathrm{i}}{\hbar}\langle[\hat{H}, \hat{p}]\rangle = -\langle \nabla V(\hat{x})\rangle \approx -V'\left(\langle\hat{x}\rangle\right)

$$


:::{admonition} Ehrenfest Equations
:class: important

These are the **Ehrenfest equations**, with the same form as Newton's second law:

$$


\frac{\mathrm{d}\langle\hat{x}\rangle}{dt} = \frac{1}{m}\langle\hat{p}\rangle

$$


$$

\frac{\mathrm{d}\langle\hat{p}\rangle}{dt} = -V'\left(\langle\hat{x}\rangle\right)

$$


**Quantum expectations of position and momentum obey classical equations of motion, on average.** This is the foundation of the **correspondence principle**.
:::

:::{admonition} Detailed Derivation
:class: dropdown note

For position:

$$


[\hat{H}, \hat{x}] = \left[\frac{\hat{p}^2}{2m} + V(\hat{x}), \hat{x}\right] = \frac{1}{2m}[\hat{p}^2, \hat{x}]

$$


Using $[\hat{p}, \hat{x}] = -\mathrm{i}\hbar$, we have

$$


[\hat{p}^2, \hat{x}] = \hat{p}[\hat{p}, \hat{x}] + [\hat{p}, \hat{x}]\hat{p} = -\mathrm{i}\hbar(\hat{p} + \hat{p}) = -2\mathrm{i}\hbar\hat{p}

$$


Thus $[\hat{H}, \hat{x}] = \frac{1}{2m}(-2\mathrm{i}\hbar\hat{p}) = -\frac{\mathrm{i}\hbar}{m}\hat{p}$, so

$$


\frac{\mathrm{d}\langle\hat{x}\rangle}{dt} = \frac{\mathrm{i}}{\hbar}\langle[\hat{H}, \hat{x}]\rangle = \frac{1}{m}\langle\hat{p}\rangle

$$


For momentum:

$$


[\hat{H}, \hat{p}] = [V(\hat{x}), \hat{p}] = -\mathrm{i}\hbar \frac{\mathrm{d}V}{\mathrm{d}\hat{x}}

$$


taking the derivative with respect to the operator $\hat{x}$. Thus

$$


\frac{\mathrm{d}\langle\hat{p}\rangle}{dt} = \frac{\mathrm{i}}{\hbar}\langle[\hat{H}, \hat{p}]\rangle = -\left\langle \frac{\mathrm{d}V}{\mathrm{d}\hat{x}}\right\rangle \approx -V'\left(\langle\hat{x}\rangle\right)

$$


in the limit where the wavefunction is well-localized.
:::

:::{admonition} Discussion: Classical Limit
:class: dropdown warning

**Does Ehrenfest's theorem imply that quantum mechanics reduces to classical mechanics for large systems?** The Ehrenfest equations look identical to Newton's laws. This suggests that quantum expectation values should follow classical trajectories. But there is a crucial catch: the validity of the approximation $\langle\nabla V(\hat{x})\rangle \approx V'(\langle\hat{x}\rangle)$ breaks down when the wavefunction spreads or becomes significantly multi-modal. For a well-localized wavefunction, the quantum and classical predictions agree; but wave packets spread. So in what regime, if any, does quantum mechanics truly "reduce" to classical mechanics?
:::

### Example: Spin Precession in the Heisenberg Picture

To see the Heisenberg picture in action, consider a spin in a static magnetic field with Hamiltonian

$$


\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z

$$


We ask: what is the time-evolution of the Pauli operators $\hat{\sigma}^x(t)$, $\hat{\sigma}^y(t)$, $\hat{\sigma}^z(t)$ in the Heisenberg picture?

#### Computing the Operator Evolution

By the Heisenberg equation,

$$


\frac{\mathrm{d}\hat{\sigma}^z}{dt} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{\sigma}^z] = \frac{\mathrm{i}\omega_0}{2}[\hat{\sigma}^z, \hat{\sigma}^z] = 0

$$


So $\hat{\sigma}^z$ is a constant of motion (as expected—the field is along z, so there is no torque in the z-direction).

For $\hat{\sigma}^x$:

$$


\frac{\mathrm{d}\hat{\sigma}^x}{dt} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{\sigma}^x] = \frac{\mathrm{i}\omega_0}{2}[\hat{\sigma}^z, \hat{\sigma}^x]

$$


Using the commutation relation $[\hat{\sigma}^z, \hat{\sigma}^x] = 2\mathrm{i}\hat{\sigma}^y$, we get

$$


\frac{\mathrm{d}\hat{\sigma}^x}{dt} = \frac{\mathrm{i}\omega_0}{2} \cdot 2\mathrm{i}\hat{\sigma}^y = -\omega_0 \hat{\sigma}^y

$$


Similarly,

$$


\frac{\mathrm{d}\hat{\sigma}^y}{dt} = \frac{\mathrm{i}\omega_0}{2}[\hat{\sigma}^z, \hat{\sigma}^y] = \frac{\mathrm{i}\omega_0}{2} \cdot (-2\mathrm{i}\hat{\sigma}^x) = \omega_0 \hat{\sigma}^x

$$


#### Solution: Larmor Precession in the Heisenberg Picture

These coupled equations have the solution

:::{admonition} Heisenberg Pauli Operators
:class: important

$$


\hat{\sigma}^x(t) = \hat{\sigma}^x(0) \cos(\omega_0 t) - \hat{\sigma}^y(0) \sin(\omega_0 t)

$$


$$

\hat{\sigma}^y(t) = \hat{\sigma}^x(0) \sin(\omega_0 t) + \hat{\sigma}^y(0) \cos(\omega_0 t)

$$


$$

\hat{\sigma}^z(t) = \hat{\sigma}^z(0)

$$


**The operators precess around the z-axis at frequency $\omega_0$**, just as a classical magnetic moment would.
:::

#### Physical Interpretation

If we start in a spin state $|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ (spin pointing in the x-direction), then

:::{admonition} Note
:class: note

$$


\langle\hat{\sigma}^x(t)\rangle = \cos(\omega_0 t), \quad \langle\hat{\sigma}^y(t)\rangle = \sin(\omega_0 t), \quad \langle\hat{\sigma}^z(t)\rangle = 0

$$


This is **Larmor precession** viewed from the Heisenberg picture, where the operators rotate while the quantum state remains fixed. The same physics emerges as in the Schrödinger picture, where states rotate and operators are fixed.
:::

### Summary

- The **Heisenberg picture** is equivalent to the Schrödinger picture but exchanges the roles of states and operators.
- **Heisenberg equation of motion**: $\frac{\mathrm{d}\hat{A}_H}{dt} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{A}_H]$
- **Conserved quantities** are those operators that commute with the Hamiltonian: $[\hat{A}, \hat{H}] = 0$.
- The **Ehrenfest theorem** shows that quantum expectations obey classical equations of motion, connecting quantum mechanics to classical mechanics.
- **Spin precession** provides a concrete example: in a magnetic field $\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z$, the Pauli operators rotate at the Larmor frequency $\omega_0$.
