## Lecture Notes

So far, we have described quantum mechanics in the **Schrödinger picture**, where states evolve in time and observables remain fixed. This is intuitive for beginners, but there is a mathematically equivalent description—the **Heisenberg picture**—where operators evolve instead of states. This perspective is more natural when discussing conservation laws and conserved quantities, and it reveals deep connections between quantum and classical mechanics. In this section, we explore operator dynamics and meet the Heisenberg equation of motion.

### The Two Pictures: Schrödinger vs. Heisenberg

In the **Schrödinger picture** (which we have used throughout this chapter):
- States $|\psi(t)angle_S$ evolve according to the Schrödinger equation: $i\hbar \frac{d}{dt}|\psi(t)angle_S = \hat{H}|\psi(t)angle_S$
- Observables $\hat{A}_S$ are time-independent
- Expectation values: $\langle A(t) angle = {}_S\langle\psi(t)|\hat{A}_S|\psi(t)angle_S$

In the **Heisenberg picture**:
- States $|\psiangle_H$ are fixed (we choose the initial state)
- Observables $\hat{A}_H(t)$ evolve in time
- Expectation values: $\langle A(t) angle = {}_H\langle\psi|\hat{A}_H(t)|\psiangle_H$

Both pictures give the **same expectation values** and observable predictions. They are related by a unitary transformation: if $U(t) = e^{-i\hat{H}t/\hbar}$, then

$$

\hat{A}_H(t) = U^\dagger(t) \hat{A}_S U(t) \quad 	ext{and} \quad |\psiangle_H = U^\dagger(t) |\psi(t)angle_S

$$

**When is each useful?**
- **Schrödinger picture**: Easier for solving time-dependent problems with explicit states; natural for computing populations and transition probabilities.
- **Heisenberg picture**: Better for identifying conserved quantities and symmetries; more closely mirrors classical mechanics; essential for understanding the correspondence principle.

### The Heisenberg Equation of Motion

To find how operators evolve in the Heisenberg picture, we start with the relation $\hat{A}_H(t) = U^\dagger(t) \hat{A}_S U(t)$ and differentiate with respect to time:

$$

\frac{d\hat{A}_H}{dt} = \frac{d U^\dagger}{dt} \hat{A}_S U + U^\dagger \hat{A}_S \frac{dU}{dt}

$$

Using $\frac{dU}{dt} = -\frac{i}{\hbar}\hat{H} U$ and $\frac{dU^\dagger}{dt} = \frac{i}{\hbar}U^\dagger \hat{H}$, we have

$$

\frac{d\hat{A}_H}{dt} = \frac{i}{\hbar}U^\dagger \hat{H} \hat{A}_S U - \frac{i}{\hbar}U^\dagger \hat{A}_S \hat{H} U = \frac{i}{\hbar}U^\dagger [\hat{H}, \hat{A}_S] U

$$

Since $\hat{H}$ commutes with both $U$ and $U^\dagger$ (it generates the evolution), this becomes

$$

\boxed{\frac{d\hat{A}_H}{dt} = \frac{i}{\hbar}[\hat{H}, \hat{A}_H] + \left(\frac{\partial \hat{A}}{\partial t}ight)_H}

$$

This is the **Heisenberg equation of motion**. The second term accounts for explicit time-dependence in the operator itself; for time-independent observables, it vanishes.

::::{admonition} Derivation: Heisenberg Equation from First Principles
:class: dropdown

We can also derive this directly. Define the Heisenberg operator as $\hat{A}_H(t) = U^\dagger(t)\hat{A}_S U(t)$. Then:

$$

\frac{d\hat{A}_H}{dt} = \left(\frac{dU^\dagger}{dt}\right)\hat{A}_S U + U^\dagger\hat{A}_S\left(\frac{dU}{dt}\right) + U^\dagger\left(\frac{\partial\hat{A}_S}{\partial t}\right)U

$$

From the Schrödinger equation, $i\hbar\dot{U} = \hat{H}U$, so $\dot{U} = -rac{i}{\hbar}\hat{H}U$ and $\dot{U}^\dagger = rac{i}{\hbar}U^\dagger\hat{H}$ (since $\hat{H}$ is Hermitian).

$$

\frac{d\hat{A}_H}{dt} = \frac{i}{\hbar}U^\dagger\hat{H}\hat{A}_S U - \frac{i}{\hbar}U^\dagger\hat{A}_S\hat{H}U + U^\dagger\left(\frac{\partial\hat{A}_S}{\partial t}\right)U

$$

$$

= \frac{i}{\hbar}U^\dagger[\hat{H}, \hat{A}_S]U + U^\dagger\left(\frac{\partial\hat{A}_S}{\partial t}\right)U = \frac{i}{\hbar}[\hat{H}, \hat{A}_H] + \left(\frac{\partial\hat{A}}{\partial t}\right)_H

$$

since $\hat{H}$ is time-independent and commutes with $U$ and $U^\dagger$.

::::

### Conserved Quantities and Conservation Laws

A key insight emerges from the Heisenberg equation: if an observable $\hat{A}$ has no explicit time-dependence and **commutes with the Hamiltonian**, then

$$

[\hat{A}, \hat{H}] = 0 \quad \Rightarrow \quad \frac{d\hat{A}_H}{dt} = 0

$$

This means $\hat{A}$ is a **constant of motion**—its expectation value is time-independent:

$$

\frac{d}{dt}\langle\hat{A}angle = \langle\hat{A}_H(t)angle - \langle\hat{A}_H(0)angle = 0

$$

**In quantum mechanics, commuting with the Hamiltonian is the precise condition for a conserved quantity.** This is the quantum mechanical expression of Noether's theorem and Wigner's symmetry principle. For every continuous symmetry of the system, there is a conserved operator.

**Examples:**
- If $[\hat{L}_z, \hat{H}] = 0$ (rotational symmetry about the z-axis), then angular momentum in the z-direction is conserved.
- If the Hamiltonian is time-independent (translational symmetry in time), then energy is conserved: $[\hat{H}, \hat{H}] = 0$ always.
- For a free particle, $[\hat{oldsymbol{p}}, \hat{H}] = 0$, so momentum is conserved.

### The Ehrenfest Theorem: Bridging Quantum and Classical

The Heisenberg equation connects quantum expectations to classical equations of motion. For a system with Hamiltonian $\hat{H} = \frac{\hat{p}^2}{2m} + V(\hat{x})$, applying the Heisenberg equation to position and momentum gives

$$

\frac{d}{dt}\langle\hat{x}angle = \frac{i}{\hbar}\langle[\hat{H}, \hat{x}]angle = \frac{1}{m}\langle\hat{p}angle

$$

$$

\frac{d}{dt}\langle\hat{p}angle = \frac{i}{\hbar}\langle[\hat{H}, \hat{p}]angle = -\langle
abla V(\hat{x})angle \approx -V'\left(\langle\hat{x}angleight)

$$

These are the **Ehrenfest equations**, and they have the same form as Newton's second law! This shows that quantum expectations of position and momentum obey classical equations of motion, on average. This is the foundation of the **correspondence principle**.

::::{admonition} Derivation: Ehrenfest Equations
:class: dropdown

For position:

$$

[\hat{H}, \hat{x}] = \left[\frac{\hat{p}^2}{2m} + V(\hat{x}), \hat{x}\right] = \frac{1}{2m}[\hat{p}^2, \hat{x}]

$$

Using $[\hat{p}, \hat{x}] = -i\hbar$, we have

$$

[\hat{p}^2, \hat{x}] = \hat{p}[\hat{p}, \hat{x}] + [\hat{p}, \hat{x}]\hat{p} = -i\hbar(\hat{p} + \hat{p}) = -2i\hbar\hat{p}

$$

Thus $[\hat{H}, \hat{x}] = \frac{1}{2m}(-2i\hbar\hat{p}) = -\frac{i\hbar}{m}\hat{p}$, so

$$

\frac{d\langle\hat{x}angle}{dt} = \frac{i}{\hbar}\langle[\hat{H}, \hat{x}]angle = \frac{1}{m}\langle\hat{p}angle

$$

For momentum:

$$

[\hat{H}, \hat{p}] = [V(\hat{x}), \hat{p}] = -i\hbar \frac{dV}{d\hat{x}}

$$

taking the derivative with respect to the operator $\hat{x}$. Thus

$$

\frac{d\langle\hat{p}angle}{dt} = \frac{i}{\hbar}\langle[\hat{H}, \hat{p}]angle = -\left\langle \frac{dV}{d\hat{x}}ightangle \approx -V'\left(\langle\hat{x}angleight)

$$

in the limit where the wavefunction is well-localized.

::::

### Example: Spin Precession in the Heisenberg Picture

To see the Heisenberg picture in action, consider a spin in a static magnetic field with Hamiltonian

$$

\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z

$$

We ask: what is the time-evolution of the Pauli operators $\hat{\sigma}^x(t)$, $\hat{\sigma}^y(t)$, $\hat{\sigma}^z(t)$ in the Heisenberg picture?

By the Heisenberg equation,

$$

\frac{d\hat{\sigma}^z}{dt} = \frac{i}{\hbar}[\hat{H}, \hat{\sigma}^z] = \frac{i\omega_0}{2}[\hat{\sigma}^z, \hat{\sigma}^z] = 0

$$

So $\hat{\sigma}^z$ is a constant of motion (as expected—the field is along z, so there is no torque in the z-direction).

For $\hat{\sigma}^x$:

$$

\frac{d\hat{\sigma}^x}{dt} = \frac{i}{\hbar}[\hat{H}, \hat{\sigma}^x] = \frac{i\omega_0}{2}[\hat{\sigma}^z, \hat{\sigma}^x]

$$

Using the commutation relation $[\hat{\sigma}^z, \hat{\sigma}^x] = 2i\hat{\sigma}^y$, we get

$$

\frac{d\hat{\sigma}^x}{dt} = \frac{i\omega_0}{2} \cdot 2i\hat{\sigma}^y = -\omega_0 \hat{\sigma}^y

$$

Similarly,

$$

\frac{d\hat{\sigma}^y}{dt} = \frac{i\omega_0}{2}[\hat{\sigma}^z, \hat{\sigma}^y] = \frac{i\omega_0}{2} \cdot (-2i\hat{\sigma}^x) = \omega_0 \hat{\sigma}^x

$$

These coupled equations have the solution

$$

\hat{\sigma}^x(t) = \hat{\sigma}^x(0) \cos(\omega_0 t) - \hat{\sigma}^y(0) \sin(\omega_0 t)

$$

$$

\hat{\sigma}^y(t) = \hat{\sigma}^x(0) \sin(\omega_0 t) + \hat{\sigma}^y(0) \cos(\omega_0 t)

$$

$$

\hat{\sigma}^z(t) = \hat{\sigma}^z(0)

$$

Remarkably, **the operators precess around the z-axis at frequency $\omega_0$**, just as a classical magnetic moment would. If we start in a spin state $|\psiangle = (|0angle + |1angle)/\sqrt{2}$ (spin pointing in the x-direction), then

$$

\langle\hat{\sigma}^x(t)angle = \cos(\omega_0 t), \quad \langle\hat{\sigma}^y(t)angle = \sin(\omega_0 t), \quad \langle\hat{\sigma}^z(t)angle = 0

$$

This is **Larmor precession** viewed from the Heisenberg picture, where the operators rotate while the quantum state remains fixed.

### Summary

- The **Heisenberg picture** is equivalent to the Schrödinger picture but exchanges the roles of states and operators.
- **Heisenberg equation of motion**: $\frac{d\hat{A}_H}{dt} = \frac{i}{\hbar}[\hat{H}, \hat{A}_H]$
- **Conserved quantities** are those operators that commute with the Hamiltonian: $[\hat{A}, \hat{H}] = 0$.
- The **Ehrenfest theorem** shows that quantum expectations obey classical equations of motion, connecting quantum mechanics to classical mechanics.
- **Spin precession** provides a concrete example: in a magnetic field $\hat{H} = \frac{\hbar\omega_0}{2}\hat{\sigma}^z$, the Pauli operators rotate at the Larmor frequency $\omega_0$.
