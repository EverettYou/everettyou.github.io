# 1.3 Time Evolution

### Overview

This chapter explores how quantum states evolve in time under the influence of a Hamiltonian. The Schrödinger equation is the fundamental equation of quantum mechanics, as central as Newton's laws are to classical mechanics.

**In quantum computing:** Quantum gates are implemented by controlled evolution under tailored Hamiltonians. A $\pi/2$-rotation (Hadamard gate) creates superposition; a $\pi$-rotation (NOT gate) flips qubits. Understanding time evolution is essential for quantum control.

**In atomic and molecular physics:** Electronic transitions, spectroscopy, and chemical reactions all involve time-dependent evolution of quantum systems.

**In fundamental physics:** The Schrödinger equation is arguably the most important equation in physics. All phenomena, from atoms to stars, ultimately obey this fundamental law (though in realistic systems, we also must account for interactions with the environment, leading to decoherence).



### Key Concepts

- **Schrödinger equation**: $i\hbar \frac{\partial|\psi\rangle}{\partial t} = \hat{H}|\psi\rangle$, connecting time evolution to the Hamiltonian (energy operator).

- **Time-evolution operator**: $\hat{U}(t) = e^{-i\hat{H}t/\hbar}$ is a **unitary operator** that directly solves the Schrödinger equation. Unitarity ensures normalization and probability conservation.

- **Stationary states**: Energy eigenstates $|E_n\rangle$ are stationary—they evolve only by a global phase $e^{-iE_n t/\hbar}$, which is unobservable. Physical quantities remain constant.

- **Superpositions and interference**: Superpositions of energy eigenstates exhibit quantum interference due to different phase evolution rates. This produces oscillatory behavior, including Rabi oscillations.

- **Larmor precession**: A spin in a magnetic field precesses around the field direction at the **Larmor frequency** $\omega_0 = \gamma B_0$. The Bloch vector rotates around the field axis.

- **Rabi oscillations**: When a two-level system is driven by an oscillating field (resonant with the transition), the population oscillates periodically between the two levels. The **Rabi frequency** $\Omega$ determines the oscillation rate.

- **Heisenberg picture**: Alternative formulation where operators evolve and states are static. The Heisenberg equation $\frac{d\hat{O}}{dt} = \frac{i}{\hbar}[\hat{H}, \hat{O}]$ governs operator dynamics.

- **Ehrenfest theorem**: Expectation values follow classical equations of motion. This is the link between quantum and classical mechanics.


### Learning Objectives

By the end of this chapter, you should be able to:
- State and solve the time-dependent Schrödinger equation
- Understand why the time-evolution operator must be unitary and what that ensures
- Recognize stationary states and explain why they don't evolve physically
- Compute the time evolution of superpositions of energy eigenstates
- Explain Larmor precession: a spin's Bloch vector rotates around the magnetic field direction
- Understand Rabi oscillations and compute transition probabilities
- Apply the Rabi formula to design quantum gates (e.g., $\pi/2$-pulse, $\pi$-pulse)
- Distinguish between the Schrödinger and Heisenberg pictures and use both
- Apply the Heisenberg equation to find conserved quantities and operator time-dependence
- State and use the Ehrenfest theorem to connect quantum to classical dynamics
