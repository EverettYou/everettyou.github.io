## Lecture Notes

Just as classical information is encoded in bit strings, quantum information is encoded in state vectors that live in a complex vector space. This section introduces the mathematical language of quantum mechanicsâ€”the Hilbert space formalismâ€”and shows how different representations of a quantum state reveal different aspects of physical reality.

:::{admonition} Quantum State: Foundation of Predictions
:class: important

In quantum mechanics, a **quantum state** is a mathematical object that encodes all information about a system necessary to predict measurement outcomes. For a qubit, the state lives in a **two-dimensional complex Hilbert space**, denoted $\mathcal{H} = \mathbb{C}^2$.

**Ket notation**: We denote a state by the **ket** $|\psi\rangle$ (read "psi-ket"). This is an abstract vectorâ€”it has no particular representation yet. Mathematically, $|\psi\rangle$ is an element of the Hilbert space.
:::

#### Properties of Quantum States

1. **Superposition principle**: If $|\psi\rangle$ and $|\phi\rangle$ are states, then any linear combination $\alpha|\psi\rangle + \beta|\phi\rangle$ (with $\alpha, \beta \in \mathbb{C}$) is also a valid state.

2. **Normalization**: We restrict to **normalized states** with $\langle\psi|\psi\rangle = 1$. This ensures probability conservation.

3. **Phase ambiguity**: Two states $|\psi\rangle$ and $\mathrm{e}^{\mathrm{i}\theta}|\psi\rangle$ (differing by a global phase) represent the **same physical state**. Global phases are unobservable.

:::{admonition} Discussion
:class: dropdown warning

**Is global phase really unobservable?** We assert that $|\psi\rangle$ and $\mathrm{e}^{\mathrm{i}\theta}|\psi\rangle$ are "the same physical state." But what if there existed an exotic measurement that compared the phase of a quantum state to some external reference? Would global phase become observable? How would we ever know if such a measurement is impossible in principle (like Lorentz invariance in relativity) versus merely impossible with current technology?

:::

#### Two-State System for a Qubit

A qubit is defined by the existence of a distinguished **computational basis**â€”two orthonormal states we call $|0\rangle$ and $|1\rangle$ (also written as $|0\rangle$ and $|1\rangle$ for spin).

**Orthonormality**:

$$


\langle 0 | 0 \rangle = 1, \quad \langle 1 | 1 \rangle = 1, \quad \langle 0 | 1 \rangle = 0

$$ (eq-orthonormal-computational)

**General superposition**:

$$


|\psi\rangle = \alpha |0\rangle + \beta |1\rangle

$$ (eq-general-qubit-state)

where $\alpha, \beta \in \mathbb{C}$ and the **normalization condition** is

$$


|\alpha|^2 + |\beta|^2 = 1

$$ (eq-normalization-condition)

### Basis Representation

While the abstract state $|\psi\rangle$ is basis-independent, once we choose a basis we can write it as a **column vector** (or **state vector**).

:::{admonition} Discussion
:class: dropdown warning

**Basis-independence: abstract freedom vs. experimental constraint**

We claim the state $|\psi\rangle$ is "basis-independent"â€”the same physical state regardless of which basis we use to describe it. Yet in any actual experiment, we must choose which basis to measure.

- Write the state $|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ in *three different bases*: computational ($\{|0\rangle, |1\rangle\}$), x-basis ($\{|+\rangle, |-\rangle\}$), and y-basis ($\{|{\mathrm{i}}\rangle, |{\bar{\mathrm{i}}}\rangle\}$). Show that the representation changes but the probabilities for measurement outcomes are consistent.
- If we measure this state in the computational basis, we get outcomes 0 and 1 with equal probability. If we measure in the x-basis, what probabilities do we get? Explain why the choice of measurement basis changes the predicted outcome probabilities, even though the state itself is unchanged.
- Does "basis-independence" mean the abstract state is somehow special or fundamental? Or is it just saying that the *mathematics* of Hilbert space is coordinate-free, while measurement always forces a concrete choice?

:::

#### Computational Basis (Z-basis)

In the computational basis $\{|0\rangle, |1\rangle\}$:

$$


|0\rangle \cong \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \quad
|1\rangle \cong \begin{pmatrix} 0 \\ 1 \end{pmatrix}, \quad
|\psi\rangle \cong \begin{pmatrix} \alpha \\ \beta \end{pmatrix}

$$ (eq-computational-representation)

The symbol $\cong$ means "is represented as"â€”the representation depends on the basis.

**Inner product in matrix form**:

$$


\langle \psi | \phi \rangle = \begin{pmatrix} \alpha^* & \beta^* \end{pmatrix} \begin{pmatrix} \gamma \\ \delta \end{pmatrix} = \alpha^* \gamma + \beta^* \delta

$$ (eq-inner-product-matrix)

where $|\phi\rangle \cong \begin{pmatrix} \gamma \\ \delta \end{pmatrix}$.

:::{admonition} Key Insight: Basis Dependence
:class: important

The abstract state $|\psi\rangle$ is **unique and basis-independent**. But the vector representation $\begin{pmatrix} \alpha \\ \beta \end{pmatrix}$ **depends on the choice of basis**. If we rotate to a different basis, the components change, but the state $|\psi\rangle$ remains the same.

This is exactly like vectors in 3D space: a vector pointing north has different components in Cartesian vs. spherical coordinates, but it is the same geometric object.
:::

### Degrees of Freedom and Parameter Counting

How many independent real parameters describe a qubit state?

**Naive count:**
- Two complex numbers $\alpha, \beta$: $2 \times 2 = 4$ real parameters (real and imaginary parts of each).

**Constraints:**
- Normalization: $|\alpha|^2 + |\beta|^2 = 1$ removes **1 real constraint**.
- Global phase: $|\psi\rangle$ and $\mathrm{e}^{\mathrm{i}\theta}|\psi\rangle$ are physically identical, removing **1 real parameter** (the global phase $\theta$).

**Effective count:** $4 - 1 - 1 = 2$ independent real parameters.

:::{admonition} Qubit Parametrization
:class: important

A qubit state can be written uniquely as

$$


|\psi\rangle = \cos(\theta/2) |0\rangle + \mathrm{e}^{\mathrm{i}\phi} \sin(\theta/2) |1\rangle

$$ (eq-bloch-parametrization-detailed)

where $\theta \in [0, \pi]$ and $\phi \in [0, 2\pi)$ are two real angles.

**Verification of normalization:**

$$


\cos^2(\theta/2) + \sin^2(\theta/2) = 1 \quad \checkmark

$$


:::

:::{admonition} Example: State Parametrization
:class: example

1. **State $|0\rangle$** (north pole): $\theta = 0, \phi$ arbitrary
   $$|\psi\rangle = \cos(0) |0\rangle + \mathrm{e}^{\mathrm{i}\phi} \sin(0) |1\rangle = |0\rangle$$


2. **State $|1\rangle$** (south pole): $\theta = \pi, \phi$ arbitrary
   $$|\psi\rangle = \cos(\pi/2) |0\rangle + \mathrm{e}^{\mathrm{i}\phi} \sin(\pi/2) |1\rangle = \mathrm{e}^{\mathrm{i}\phi} |1\rangle \approx |1\rangle$$


3. **State $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$** (positive x): $\theta = \pi/2, \phi = 0$
   $$|\psi\rangle = \cos(\pi/4) |0\rangle + \mathrm{e}^{\mathrm{i} \cdot 0} \sin(\pi/4) |1\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) \quad \checkmark$$


4. **State $|\mathrm{i}\rangle = \frac{1}{\sqrt{2}}(|0\rangle + \mathrm{i}|1\rangle)$**: $\theta = \pi/2, \phi = \pi/2$
   $$|\psi\rangle = \cos(\pi/4) |0\rangle + \mathrm{e}^{\mathrm{i}\pi/2} \sin(\pi/4) |1\rangle = \frac{1}{\sqrt{2}}(|0\rangle + \mathrm{i}|1\rangle) \quad \checkmark$$


:::

### Basis Transformations

The choice of basis is arbitrary. Any two orthonormal bases are related by a **unitary transformation**.

#### X-basis

The **x-basis** (eigenstates of $\hat{\sigma}^x$) is

$$


|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \quad
|-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)

$$ (eq-x-basis)

These states are orthonormal.

:::{admonition} Derivation: Orthonormality of X-basis
:class: dropdown note

$$


\langle + | + \rangle = \frac{1}{2}(\langle 0| + \langle 1|)(|0\rangle + |1\rangle) = \frac{1}{2}(1 + 0 + 0 + 1) = 1 \quad \checkmark

$$


$$

\langle + | - \rangle = \frac{1}{2}(\langle 0| + \langle 1|)(|0\rangle - |1\rangle) = \frac{1}{2}(1 - 0 + 0 - 1) = 0 \quad \checkmark

$$


:::

**Inverse relations** (expressing computational basis in terms of x-basis):

$$


|0\rangle = \frac{1}{\sqrt{2}}(|+\rangle + |-\rangle), \quad
|1\rangle = \frac{1}{\sqrt{2}}(|+\rangle - |-\rangle)

$$ (eq-z-in-x-basis)

#### Y-basis

The **y-basis** (eigenstates of $\hat{\sigma}^y$) is

$$


|{\mathrm{i}}\rangle = \frac{1}{\sqrt{2}}(|0\rangle + \mathrm{i}|1\rangle), \quad
|{\bar{\mathrm{i}}}\rangle = \frac{1}{\sqrt{2}}(|0\rangle - \mathrm{i}|1\rangle)

$$ (eq-y-basis)

### The Bloch Sphere

:::{margin}
:::{figure} ../assets/figs/1-1-2-block_sphere.png
:name: fig-bloch-sphere
:width: 100%

The Bloch sphere: pure qubit states $|\psi\rangle$ map to points on $S^2$; polar angle $\theta$ and azimuth $\phi$ match the standard parametrization; basis kets lie at the poles and equator as labeled.
::::::

The **Bloch sphere** is a geometric representation of the space of single-qubit pure states. Every normalized qubit state corresponds to a unique point on the unit sphere in 3D.

#### Parametrization and Geometry

Using spherical coordinates:

$$


|\psi\rangle = \cos(\theta/2) |0\rangle + \mathrm{e}^{\mathrm{i}\phi} \sin(\theta/2) |1\rangle

$$ (eq-bloch-sphere-param)

where:
- $\theta \in [0, \pi]$: **polar angle** from the north pole
- $\phi \in [0, 2\pi)$: **azimuthal angle** around the z-axis

These two angles are sufficient to specify any pure qubit state.

#### Key Points on the Bloch Sphere

| State | $\theta$ | $\phi$ | Position |
|-------|----------|--------|----------|
| $\|0\rangle$ | 0 | any | North pole |
| $\|1\rangle$ | $\pi$ | any | South pole |
| $\|+\rangle = \frac{1}{\sqrt{2}}(\|0\rangle + \|1\rangle)$ | $\pi/2$ | 0 | Positive x-axis |
| $\|-\rangle = \frac{1}{\sqrt{2}}(\|0\rangle - \|1\rangle)$ | $\pi/2$ | $\pi$ | Negative x-axis |
| $\|{\mathrm{i}}\rangle = \frac{1}{\sqrt{2}}(\|0\rangle + i\|1\rangle)$ | $\pi/2$ | $\pi/2$ | Positive y-axis |
| $\|{\bar{\mathrm{i}}}\rangle = \frac{1}{\sqrt{2}}(\|0\rangle - i\|1\rangle)$ | $\pi/2$ | $3\pi/2$ | Negative y-axis |

#### Connection to Spin Expectation Values

For a qubit state $|\psi\rangle$, the **expectation values** of the Pauli operators form a 3D vector:

$$


\boldsymbol{n} = \langle\psi|\hat{\boldsymbol{\sigma}}|\psi\rangle = (\langle \hat{\sigma}^x \rangle, \langle \hat{\sigma}^y \rangle, \langle \hat{\sigma}^z \rangle) = (\sin\theta \cos\phi, \sin\theta \sin\phi, \cos\theta)

$$ (eq-bloch-vector)

This is a **unit vector** pointing from the origin to the location of $|\psi\rangle$ on the Bloch sphere.


:::{admonition} Discussion
:class: dropdown warning

**The Bloch sphere has no generalization to qutrit (3-level) systems â€” why?**

A qubit state is characterized by 2 real parameters (after removing global phase and normalization). These map to the surface of a 3D sphere $S^2$.

- Count the real parameters needed to specify a pure state of a qutrit (3-level system). After removing overall phase and normalization, how many free real parameters remain?
- For a qutrit, can you find a geometric space (like a sphere) that parametrizes all pure states? What is its dimension? (Hint: look up the "complex projective plane" $\mathbb{CP}^2$.)
- The Bloch sphere is particularly useful because rotations in 3D space (the $SU(2)$ symmetry group) directly correspond to quantum gates acting on the qubit. Does a similar geometric picture hold for the qutrit, or is the special role of $S^2$ unique to qubits?
- Measurement of a qubit gives 2 outcomes; measurement of a qutrit gives 3. How does the measurement formalism change? Can you still represent a qutrit state on a simple sphere?

:::

:::{admonition} Geometric Interpretation
:class: important

The Bloch sphere provides a complete geometric picture of all pure qubit states:
- **North pole** ($|0\rangle$): Maximum z-spin, zero x and y spin.
- **South pole** ($|1\rangle$): Opposite z-spin.
- **Equator**: Superpositions with equal amplitudes in $|0\rangle$ and $|1\rangle$; zero z-expectation, nonzero x and y.
- **Interior points**: Not reachable by pure states (reserved for mixed states).
:::

:::{admonition} Born Rule: Amplitudes to Probabilities
:class: important

The coefficients $\alpha$ and $\beta$ in the superposition

$$


|\psi\rangle = \alpha |0\rangle + \beta |1\rangle

$$


have a direct **probabilistic interpretation**:

**If we measure the qubit in the computational basis** $\{|0\rangle, |1\rangle\}$:
- Outcome 0 occurs with probability $|\alpha|^2$
- Outcome 1 occurs with probability $|\beta|^2$
- The **normalization condition** $|\alpha|^2 + |\beta|^2 = 1$ ensures that probabilities sum to 1.

**Important:** The phases of $\alpha$ and $\beta$ matter for predictions about other measurements, but they do not affect the probabilities for computational-basis measurements.
:::

### Summary

- A **qubit state** is an abstract vector $|\psi\rangle$ in a 2D complex Hilbert space.
- **Basis representation** expresses the state as a column vector; the representation depends on the choice of basis.
- A general qubit state is parametrized by **2 real angles** ($\theta$, $\phi$) on the **Bloch sphere**.
- **Basis transformations** relate different representations of the same state; unitary transformations rotate the state on the Bloch sphere.
- The **Born rule** connects the amplitudes $\alpha$ and $\beta$ to measurement probabilities.
