## Lecture Notes

### Overview

Quantum mechanics assigns to each physical system a **state vector** $\vert \psi\rangle$ that encodes all information necessary to predict measurement outcomes. This section develops the mathematical language—kets, bras, inner products, and basis representations—and introduces the **Bloch sphere**, a geometric visualization of the space of qubit states.

### Ket, Bra, and Inner Product

**Ket notation** represents a quantum state as an abstract vector in a complex Hilbert space. For a two-level system (qubit), the state lives in $\mathcal{H} = \mathbb{C}^2$.

:::{admonition} Definition: Ket and Bra
:class: important

- The **ket** $\vert \psi\rangle$ is an abstract vector in Hilbert space.
- The **bra** $\langle \psi\vert $ is the dual (Hermitian conjugate) of the ket: $\langle \psi\vert = (\vert \psi\rangle)^\dagger$.
- The **inner product** $\langle\phi\vert \psi\rangle$ is a complex number defined by:

$$

\langle\phi\vert \psi\rangle \equiv \sum_i (\phi_i)^* \psi_i

$$

where $\phi_i$ and $\psi_i$ are components in some basis.

- **Normalization**: A state is **normalized** if $\langle\psi\vert \psi\rangle = 1$.
- **Orthogonality**: Two states are **orthogonal** if $\langle\phi\vert \psi\rangle = 0$.

:::

**In vector form**, once we choose a basis:

| Object | Notation | Type | Representation |
|--------|----------|------|----------------|
| **Ket** (state) | $\vert\psi\rangle$ | Column vector | $\begin{pmatrix} \psi_1 \\ \psi_2 \end{pmatrix}$ |
| **Bra** (dual) | $\langle\psi\vert$ | Row vector | $(\psi_1^*\ \ \psi_2^*)$ |
| **Inner product** | $\langle\phi\vert\psi\rangle$ | Scalar | $\phi_1^*\psi_1 + \phi_2^*\psi_2$ |
| **Outer product** | $\vert\psi\rangle\langle\phi\vert$ | Matrix | $\begin{pmatrix} \psi_1\phi_1^* & \psi_1\phi_2^* \\ \psi_2\phi_1^* & \psi_2\phi_2^* \end{pmatrix}$ |

### The Qubit State

A qubit is defined by the existence of two **orthonormal computational basis states** $\vert 0\rangle$ and $\vert 1\rangle$:

$$

\langle 0\vert 0\rangle = 1, \quad \langle 1\vert 1\rangle = 1, \quad \langle 0\vert 1\rangle = 0

$$ (eq-orthonormal-computational)

**Any qubit state** can be written as a superposition of these basis states:

$$

|\psi\rangle = \alpha |0\rangle + \beta |1\rangle

$$ (eq-general-qubit-state)

where $\alpha, \beta \in \mathbb{C}$ are **amplitudes** and the **normalization condition** is

$$

|\alpha|^2 + |\beta|^2 = 1

$$ (eq-normalization-condition)

:::{admonition} State vs. Representation
:class: note

Strictly speaking, $\vert \psi\rangle$ is an abstract vector while $\begin{pmatrix} \alpha \\ \beta \end{pmatrix}$ is its representation in a chosen basis. Throughout these notes we write $\vert \psi\rangle = \begin{pmatrix} \alpha \\ \beta \end{pmatrix}$ using the equality sign, with the understanding that the right-hand side **depends on the basis**. If we switch to a different basis, the components change even though the state $\vert \psi\rangle$ is the same—just as a 3D vector has different components in Cartesian vs. spherical coordinates. The same remark applies to operators and their matrix representations.
:::

**Conversion between ket and vector notation**: To express $\vert \psi\rangle = \alpha \vert 0\rangle + \beta \vert 1\rangle$ as components, write $(\alpha, \beta)$. To go the other way, interpret components $(\alpha, \beta)$ as the ket $\alpha \vert 0\rangle + \beta \vert 1\rangle$.

### Born's Rule and Normalization

When we measure a qubit in the computational basis, the **Born rule** gives the measurement probabilities:

:::{admonition} Born Rule
:class: important

If $\vert \psi\rangle = \alpha \vert 0\rangle + \beta \vert 1\rangle$, then measurement of the qubit in the computational basis $\{\vert 0\rangle, \vert 1\rangle\}$ yields:

$$

P(\text{outcome } 0) = |\alpha|^2, \quad P(\text{outcome } 1) = |\beta|^2

$$

The normalization condition $|\alpha|^2 + |\beta|^2 = 1$ ensures probabilities sum to unity: $P(0) + P(1) = 1$.

:::

**Global phase is unobservable**: Two states $\vert \psi\rangle$ and $\mathrm{e}^{\mathrm{i}\gamma}\vert \psi\rangle$ (differing by a global phase $\gamma \in \mathbb{R}$) give identical probabilities for all measurements. For example:

$$

|\psi\rangle = \alpha |0\rangle + \beta |1\rangle \quad \text{vs} \quad |\psi'\rangle = \mathrm{e}^{\mathrm{i}\gamma}(\alpha |0\rangle + \beta |1\rangle)

$$

Both have $P(0) = |\alpha|^2$ and $P(1) = |\beta|^2$. Global phases do not affect physical predictions and can be freely removed.

### Parameter Counting and Bloch Parametrization

How many independent real parameters specify a qubit state?

**Naive count:** Two complex amplitudes $\alpha, \beta$ give $2 \times 2 = 4$ real parameters (real and imaginary parts).

**Constraints:**
- Normalization: $|\alpha|^2 + |\beta|^2 = 1$ removes **1 constraint**.
- Global phase: The freedom to multiply by $\mathrm{e}^{\mathrm{i}\gamma}$ removes **1 parameter**.

**Effective count:** $4 - 1 - 1 = 2$ independent real parameters.

These two parameters naturally parametrize the qubit state:

:::{admonition} Bloch Parametrization
:class: important

Every qubit state can be written uniquely as

$$

|\psi\rangle = \cos(\theta/2) |0\rangle + \mathrm{e}^{\mathrm{i}\phi} \sin(\theta/2) |1\rangle

$$ (eq-bloch-parametrization-standard)

where $\theta \in [0, \pi]$ and $\phi \in [0, 2\pi)$ are two real angles. This parametrization automatically satisfies normalization:

$$

\cos^2(\theta/2) + \sin^2(\theta/2) = 1 \quad \checkmark

$$

:::

:::{admonition} Example: Common Qubit States in Standard Form
:class: example dropdown

1. **State $\vert 0\rangle$**: $\theta = 0$, $\phi$ arbitrary
   
   $$\vert \psi\rangle = \cos(0)\vert 0\rangle + \mathrm{e}^{\mathrm{i}\phi}\sin(0)\vert 1\rangle = \vert 0\rangle$$


2. **State $\vert 1\rangle$**: $\theta = \pi$, $\phi$ arbitrary
   
   $$\vert \psi\rangle = \cos(\pi/2)\vert 0\rangle + \mathrm{e}^{\mathrm{i}\phi}\sin(\pi/2)\vert 1\rangle = \mathrm{e}^{\mathrm{i}\phi}\vert 1\rangle \cong \vert 1\rangle$$


3. **State $\vert +\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$**: $\theta = \pi/2$, $\phi = 0$
   
   $$\vert \psi\rangle = \cos(\pi/4)\vert 0\rangle + \mathrm{e}^{\mathrm{i} \cdot 0}\sin(\pi/4)\vert 1\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle) \quad \checkmark$$

:::

### The Bloch Sphere

The geometric picture of qubit states is the **Bloch sphere**: a unit 2-sphere in 3D space where each point represents a unique qubit state.

:::{margin}
:::{figure} ../assets/figs/1-1-2-block_sphere.png
:name: fig-bloch-sphere
:width: 100%

The Bloch sphere: pure qubit states map to points on the unit sphere. Polar angle $\theta$ (from north pole) and azimuthal angle $\phi$ (around z-axis) parametrize the state; basis states lie at the poles and equator.
:::
:::

In spherical coordinates, the Bloch parametrization becomes:

$$

|\psi\rangle = \cos(\theta/2) |0\rangle + \mathrm{e}^{\mathrm{i}\phi} \sin(\theta/2) |1\rangle

$$ (eq-bloch-sphere-param)

where:
- $\theta \in [0, \pi]$: **polar angle** measured from the north pole (z-axis)
- $\phi \in [0, 2\pi)$: **azimuthal angle** around the z-axis

**Key basis states on the Bloch sphere:**

| State | $\theta$ | $\phi$ | Location |
|-------|----------|--------|----------|
| $\vert 0\rangle$ | 0 | any | North pole |
| $\vert 1\rangle$ | $\pi$ | any | South pole |
| $\vert +\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$ | $\pi/2$ | 0 | Positive x-axis |
| $\vert -\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle - \vert 1\rangle)$ | $\pi/2$ | $\pi$ | Negative x-axis |
| $\vert \mathrm{i}\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \mathrm{i}\vert 1\rangle)$ | $\pi/2$ | $\pi/2$ | Positive y-axis |
| $\vert \bar{\mathrm{i}}\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle - \mathrm{i}\vert 1\rangle)$ | $\pi/2$ | $3\pi/2$ | Negative y-axis |

### Orthonormal Bases

Besides the computational (Z) basis, any orthonormal pair of states defines a valid basis. We highlight the X and Y bases:

**X-basis** (eigenstates of $\hat{\sigma}^x$):

$$

|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \quad |-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)

$$ (eq-x-basis)

**Y-basis** (eigenstates of $\hat{\sigma}^y$):

$$

|\mathrm{i}\rangle = \frac{1}{\sqrt{2}}(|0\rangle + \mathrm{i}|1\rangle), \quad |\bar{\mathrm{i}}\rangle = \frac{1}{\sqrt{2}}(|0\rangle - \mathrm{i}|1\rangle)

$$ (eq-y-basis)

The **inverse relations** write the computational kets $\vert 0\rangle$ and $\vert 1\rangle$ in the X-basis and in the Y-basis (invert Eqs. [](#eq-x-basis) and [](#eq-y-basis)):

**In the X-basis** ($\{\vert +\rangle, \vert -\rangle\}$):

$$

|0\rangle = \frac{1}{\sqrt{2}}(|+\rangle + |-\rangle), \quad |1\rangle = \frac{1}{\sqrt{2}}(|+\rangle - |-\rangle)

$$ (eq-z-in-x-basis)

**In the Y-basis** ($\{\vert \mathrm{i}\rangle, \vert \bar{\mathrm{i}}\rangle\}$):

$$

|0\rangle = \frac{1}{\sqrt{2}}(|\mathrm{i}\rangle + |\bar{\mathrm{i}}\rangle), \quad |1\rangle = -\frac{\mathrm{i}}{\sqrt{2}}(|\mathrm{i}\rangle - |\bar{\mathrm{i}}\rangle)

$$ (eq-z-in-y-basis)

Each basis corresponds to a pair of antipodal points on the Bloch sphere.

:::{admonition} Basis Change Example
:class: dropdown note

To express a state in a different basis, substitute the inverse relations from Eq. [](#eq-z-in-x-basis) or [](#eq-z-in-y-basis). For example, using the X-basis, $\vert \psi\rangle = \alpha \vert 0\rangle + \beta \vert 1\rangle$ becomes:

$$

|\psi\rangle = \frac{\alpha + \beta}{\sqrt{2}}|+\rangle + \frac{\alpha - \beta}{\sqrt{2}}|-\rangle

$$

:::

### Spin Expectation Values and the Bloch Vector

For a qubit state $\vert \psi\rangle$, the expectation values of the Pauli operators form a 3D unit vector:

:::{admonition} Bloch Vector
:class: important

For a state $\vert \psi\rangle = \cos(\theta/2)\vert 0\rangle + \mathrm{e}^{\mathrm{i}\phi}\sin(\theta/2)\vert 1\rangle$, the Bloch vector is

$$

\boldsymbol{n} = (\langle\hat{\sigma}^x\rangle, \langle\hat{\sigma}^y\rangle, \langle\hat{\sigma}^z\rangle) = (\sin\theta\cos\phi, \sin\theta\sin\phi, \cos\theta)

$$ (eq-bloch-vector)

This is a **unit vector**: $|\boldsymbol{n}| = 1$. It points from the origin directly to the state's location on the Bloch sphere.

**Physical interpretation**: The Bloch vector describes the expected outcome of measuring spin in direction $\hat{\boldsymbol{n}}$.

:::

**Note**: The Pauli operators $\hat{\sigma}^x$, $\hat{\sigma}^y$, $\hat{\sigma}^z$ are introduced formally in §1.1.3. For now, understand them as the three Hermitian observables whose eigenstates define the three orthonormal bases on the Bloch sphere.

:::{admonition} Discussion
:class: dropdown tip

**Why does the Bloch sphere fail for qubits with more than two levels?**

A qubit state is specified by 2 real parameters, which naturally maps to a 2-sphere (the surface of a 3D ball). For a qutrit (3-level system), count the real parameters: $2n$ from $n$ complex amplitudes, minus 1 for normalization, minus 1 for global phase, gives $2(3) - 2 = 4$ real parameters.

- Is there a simple geometric object (like a sphere) that parametrizes all pure qutrit states?
- How does the measurement formalism change for qubits vs. qutrits?
- The Bloch sphere is useful because rotations in 3D ($SU(2)$ symmetry) correspond to quantum gates. Does a similar structure exist for higher-dimensional systems?

:::

### Summary

- A **quantum state** $\vert \psi\rangle$ is an abstract vector in a complex Hilbert space; its **representation** as components $(\alpha, \beta)$ depends on the chosen basis.
- The **Born rule** connects amplitudes to measurement probabilities: $P(0) = |\alpha|^2$, $P(1) = |\beta|^2$. Global phase is unobservable.
- A qubit state is specified by **2 real parameters**: the Bloch angles $\theta \in [0, \pi]$ and $\phi \in [0, 2\pi)$.
- The **Bloch sphere** provides a complete geometric picture of all pure qubit states as points on a unit 2-sphere.
- Three orthonormal bases (Z, X, Y) correspond to three pairs of antipodal points on the Bloch sphere; measurement basis choice determines measurement outcome statistics.
