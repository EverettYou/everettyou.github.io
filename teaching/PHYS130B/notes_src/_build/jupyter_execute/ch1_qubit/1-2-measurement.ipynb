# 1.2 Measurement

### Overview

Measurement in quantum mechanics is fundamentally different from measurement in classical physics. In classical systems, we observe pre-existing properties. In quantum systems, measurement **creates** the outcome and **collapses** the state.

This section explores three pillars of quantum measurement:

1. **The measurement postulate**: How outcomes are predicted (Born rule), and how the state changes after measurement.
2. **Incompatibility**: Why we cannot simultaneously know the values of all observables; the uncertainty principle and complementarity.
3. **Measurement operators**: The projector formalism for measurement, state collapse as knowledge update, and the Bayesian perspective.
