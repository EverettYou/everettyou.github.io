## Lecture Notes

Fermi's golden rule distills the essence of quantum transitions: when a system couples weakly to a dense spectrum of final states, the transition probability becomes a timeless rate, linear in time and proportional to the squared coupling strength times the density of available targets. This rule underlies atomic spectroscopy, radioactive decay, and particle scattering—it is the bridge between perturbation theory's formal apparatus and the measurable transition rates that define real quantum phenomena.

### Transition Probability: First-Order Calculation

#### Definition

Consider a system initially in state $|i\rangle$ (eigenstate of $H_0$ with energy $E_i^{(0)}$). The **transition probability** to reach state $|f\rangle$ (energy $E_f^{(0)}$) at time $t$ is:

$$


P_{i\to f}(t) = |\langle f | U_S(t) | i \rangle|^2

$$ (eq-5b-prob-def)

Using the factorization $U_S(t) = \mathrm{e}^{-\mathrm{i}H_0 t/\hbar} U_I(t)$, the overall phase cancels and we get:

$$


P_{i\to f}(t) = |\langle f | U_I(t) | i \rangle|^2

$$ (eq-5b-prob-IP)

To **first order** in the coupling $\lambda$ (valid for $|V| t / \hbar \ll 1$), using the first-order term of the Dyson series:

$$


\boxed{P_{i\to f}^{(1)}(t) = \frac{1}{\hbar^2} \left| \int_0^t dt_1 \mathrm{e}^{\mathrm{i}\omega_{fi} t_1} V_{fi}(t_1) \right|^2}

$$ (eq-5b-P1st-general)

where $V_{fi}(t) = \langle f | V(t) | i \rangle$ is the matrix element, and:

$$


\omega_{fi} = \frac{E_f^{(0)} - E_i^{(0)}}{\hbar}

$$ (eq-5b-omega-fi)

is the **transition frequency** (resonance is at $\omega_{fi} = 0$, i.e., $E_f = E_i$).

### Sinc Function and Resonance

#### Time-Independent Perturbation: Constant Coupling

For a **constant (time-independent) perturbation** $V(t) = V_0$ applied for a duration $t$:

$$


\boxed{P_{i\to f}^{(1)}(t) = |V_{fi}|^2 \frac{\sin^2(\omega_{fi} t/2)}{\hbar^2 (\omega_{fi}/2)^2} = |V_{fi}|^2 t^2 \operatorname{sinc}^2\left(\frac{\omega_{fi} t}{2}\right)}

$$ (eq-5b-P-sinc-formula)

where $\operatorname{sinc}(x) = \sin(x)/x$ is the sinc function.

:::{admonition} Resonance vs. Off-Resonance
:class: note

**On resonance** ($\omega_{fi} \to 0$, i.e., $E_f \to E_i$): $P \approx |V_{fi}|^2 t^2/\hbar^2$, which grows **quadratically in time**. This implies a **constant transition rate** $\Gamma = dP/dt \approx 2|V_{fi}|^2 t/\hbar^2$ (linear in $t$).

**Off-resonance** ($|\omega_{fi}| \gg 1/t$): The sinc function oscillates rapidly and averages to a small value $\propto 1/\omega_{fi}^2$, **suppressing the transition**. The **width of the main lobe** is $\Delta \omega \sim 2\pi / t$ (resolution improves with longer observation time).

:::

:::{admonition} Derivation: The Sinc Function
:class: dropdown

For constant $V_0$:

$$


P_{i\to f}^{(1)}(t) = \frac{|V_{fi}|^2}{\hbar^2} \left| \int_0^t dt_1 \mathrm{e}^{\mathrm{i}\omega_{fi}t_1} \right|^2

$$


Evaluate the integral:

$$


\int_0^t dt_1 \mathrm{e}^{\mathrm{i}\omega_{fi}t_1} = \frac{\mathrm{e}^{\mathrm{i}\omega_{fi}t} - 1}{\mathrm{i}\omega_{fi}}

$$


For $\omega_{fi} \neq 0$:

$$


\left|\int_0^t dt_1 \mathrm{e}^{\mathrm{i}\omega_{fi}t_1}\right|^2 = \frac{(\mathrm{e}^{\mathrm{i}\omega_{fi}t} - 1)(\mathrm{e}^{-\mathrm{i}\omega_{fi}t} - 1)}{\omega_{fi}^2} = \frac{2 - 2\cos(\omega_{fi}t)}{\omega_{fi}^2} = \frac{4\sin^2(\omega_{fi}t/2)}{\omega_{fi}^2}

$$


Therefore:

$$


P_{i\to f}^{(1)}(t) = |V_{fi}|^2 \frac{\sin^2(\omega_{fi}t/2)}{\hbar^2(\omega_{fi}/2)^2}

$$


:::

**Physical picture**: The sinc function encodes the interplay of **coherence and interference**. Imagine shining a laser at a specific frequency on an atom. If the laser frequency matches the atomic transition ($\omega_{fi} = 0$), absorption is strong and builds up. If the laser is detuned, the oscillating coupling causes the atom to absorb and re-emit, with the oscillations mostly canceling out. Only transitions within a bandwidth $\Delta\omega \sim 2\pi/T$ (resolution-time product) give net probability.

### Fermi's Golden Rule

#### From Sinc to Delta Function: Long-Time Limit

In real systems, the final state is usually part of a **continuum** or **dense spectrum**: scattering into all possible momenta, decay into photons with all energies, absorption across a band of states. We do not ask "what is the probability to reach a specific state $|f\rangle$?" but rather "what is the total transition rate into any state near energy $E_i$?"

The sinc function becomes increasingly sharply peaked at $\omega_{fi} = 0$ as $t \to \infty$. In the limit:

$$


\lim_{t \to \infty} \frac{t \sin^2(\omega_{fi} t/2)}{(\omega_{fi}/2)^2} = \pi t \delta(\omega_{fi})

$$


This delta function enforces **energy conservation**: only final states with $E_f = E_i$ contribute.

#### Fermi's Golden Rule Statement

Consider the total transition probability into all final states within an energy interval:

$$


P_{i \to \text{continuum}}(t) = \sum_{f} P_{i\to f}(t) \approx \int dE_f \, \rho(E_f) P_{i\to f}^{(1)}(t)

$$


where $\rho(E_f)$ is the **density of states** (number of states per unit energy).

:::{admonition} Fermi's Golden Rule
:class: important

In the **long-time limit** $t \to \infty$, the transition rate (probability per unit time) becomes constant:

$$


\boxed{\Gamma_{i\to f} = \frac{2\pi}{\hbar} |\langle f | V | i \rangle|^2 \rho(E_i)}

$$ (eq-5b-fgr-standard)

or equivalently:

$$


\boxed{\Gamma_{i\to f} = \frac{2\pi}{\hbar} |\langle f | V | i \rangle|^2 \delta(E_f - E_i)}

$$ (eq-5b-fgr-detailed)

**Physical meaning**: The transition rate is proportional to:
1. The square of the matrix element $|\langle f | V | i \rangle|^2$ (coupling strength)
2. The density of final states $\rho(E_i)$ (available target states)
3. Energy conservation via $\delta(E_f - E_i)$ (only resonant states contribute)

:::



:::{admonition} Fermi's Golden Rule: Detailed Form
:class: important

In the **long-time limit** $t \to \infty$, when a system transitions into a continuum or dense spectrum of final states, the transition rate becomes constant and independent of time:

$$


\Gamma_{{i \to f}} = \frac{{2\pi}}{{\hbar}} |\langle f | V | i \rangle|^2 \delta(E_f - E_i)

$$


or equivalently:

$$


\Gamma_{{i \to f}} = \frac{{2\pi}}{{\hbar}} |\langle f | V | i \rangle|^2 \rho(E_i)

$$


**Three essential factors**:
1. **Coupling strength** $|\langle f | V | i \rangle|^2$: Square of the matrix element (perturbation strength)
2. **Density of final states** $\rho(E_i)$: Number of available target states at the transition energy
3. **Energy conservation** $\delta(E_f - E_i)$: Only resonant final states contribute

The rule is **universal**: it applies to atomic decay, particle scattering, photoabsorption, and any weak transition process.

:::

:::{admonition} Derivation: Golden Rule from the Delta Function Limit
:class: dropdown

Start from the sinc-function formula:

$$


P_{i\to f}^{(1)}(t) = |V_{fi}|^2 \frac{\sin^2(\omega_{fi}t/2)}{\hbar^2(\omega_{fi}/2)^2}

$$


**Sum over final states** in a continuum with density $\rho(E_f)$:

$$


P_{i\to\text{continuum}}(t) = \int dE_f \, \rho(E_f) |V_{fi}(E_f)|^2 \frac{\sin^2(\omega_{fi}t/2)}{\hbar^2(\omega_{fi}/2)^2}

$$


Change variables to $\omega_{fi} = (E_f - E_i)/\hbar$, so $dE_f = \hbar \, d\omega_{fi}$:

$$


P_{i\to\text{continuum}}(t) = \int d\omega_{fi} \, \rho(E_i + \hbar\omega_{fi}) |V_{fi}(\omega_{fi})|^2 \frac{t \sin^2(\omega_{fi}t/2)}{\hbar(\omega_{fi}/2)^2}

$$


**In the long-time limit** $t \to \infty$, the function $t \sin^2(\omega_{fi}t/2) / (\omega_{fi}/2)^2$ becomes sharply peaked around $\omega_{fi} = 0$ with width $\sim 2\pi/t \to 0$. Over this narrow range, $\rho(E_f)$ and $|V_{fi}|^2$ are essentially constant, so we can approximate:

$$


\rho(E_i + \hbar\omega_{fi}) \approx \rho(E_i), \quad |V_{fi}(\omega_{fi})|^2 \approx |V_{fi}|^2

$$


Using the delta-function identity:

$$


\lim_{t\to\infty} \frac{t\sin^2(\omega_{fi}t/2)}{(\omega_{fi}/2)^2} = \pi t \delta(\omega_{fi})

$$


we get:

$$


P_{i\to\text{continuum}}(t) = \rho(E_i) |V_{fi}|^2 \frac{\pi t}{\hbar}

$$


**Transition rate**:

$$


\Gamma = \frac{\mathrm{d}P}{dt} = \rho(E_i) |V_{fi}|^2 \frac{\pi}{\hbar}

$$


(Standard conventions use $2\pi/\hbar$.)

:::

### Applications

#### Spontaneous Emission (Radiative Decay)

An excited atomic state $|i\rangle$ decays to lower states by emitting photons. The decay rate is:

$$


\Gamma_{\text{decay}} = \frac{2\pi}{\hbar} |\langle f | \boldsymbol{d} \cdot \boldsymbol{E} | i \rangle|^2 \rho_\gamma(E_f)

$$


where $\rho_\gamma$ is the **photon density of states** (number of electromagnetic radiation modes). The **lifetime** is:

$$


\tau = \frac{1}{\Gamma}

$$


Faster radiative decay occurs at higher frequencies (shorter wavelengths) because the photon density of states $\rho_\gamma(E) \propto E^2$ increases with energy.

#### Scattering Cross Section

In particle scattering (Born approximation), the final state is a plane wave with all possible momenta:

$$


\frac{\mathrm{d}\sigma}{d\Omega} \propto |\langle\boldsymbol{k}_f | V | \boldsymbol{k}_i \rangle|^2

$$


The **differential cross section** quantifies the probability of scattering into a particular direction, weighted by the density of final plane-wave states.

#### Photoabsorption

A photon of energy $\hbar\omega$ excites an electron across a band gap. The **absorption rate** depends on:
- The density of initial states (valence band)
- The density of final states (conduction band)  
- Photon energy resonance (matches band gap)
- Dipole matrix element (coupling strength)

Fermi's rule accounts for all these factors in one formula.

### Validity Conditions and Assumptions

Fermi's golden rule applies when:

:::{admonition} Validity Conditions
:class: caution

1. **First-order approximation**: $|V| t / \hbar \ll 1$ or $\rho |V|^2 \ll 1$ (coupling much weaker than density-of-states scale).
2. **Long-time limit**: $t \gg 1/|\omega_{fi}|$ (evolution time much longer than inverse transition frequency).
3. **Continuous or dense final-state spectrum**: Final states must form a continuum or be closely spaced.
4. **No initial-final overlap**: $\langle i | f \rangle = 0$ or negligible.
5. **Energy conservation**: Final states span a range that includes $E_i$ (resonant states available).

When any of these breaks down, more refined calculations are needed (second-order perturbation, beyond Born approximation, etc.).

:::

### Selection Rules and Forbidden Transitions

#### Symmetry-Based Constraints

The matrix element $\langle f | V | i \rangle$ **vanishes identically** if $V$ and the basis states have incompatible symmetries. These are **selection rules**.

:::{admonition} Example: Electric Dipole Selection Rules
:class: example

For electric dipole radiation ($V = -\boldsymbol{d} \cdot \boldsymbol{E}$, dipole moment coupled to field):

- $\Delta \ell = \pm 1$ (orbital angular momentum must change by 1)
- $\Delta m_\ell = 0, \pm 1$ (magnetic quantum number; depends on light polarization)
- $\Delta S = 0$ (spin is not affected in lowest order; spin-flip requires spin-orbit coupling)

**Example transitions**:
- $|2s\rangle \to |1s\rangle$: **Forbidden** ($\Delta\ell = 0$)
- $|2p\rangle \to |1s\rangle$: **Allowed** ($\Delta\ell = -1$)
- $|2p, m_\ell=1\rangle \to |1s, m_\ell=0\rangle$ with linear-$z$ polarized light: **Allowed** ($\Delta m_\ell = -1$)

:::




:::{admonition} Selection Rules: Symmetry-Forbidden Transitions
:class: important

**Selection rules** are constraints on allowed transitions arising from the symmetries of the perturbation $V$ and the basis states. A matrix element $\langle f | V | i \rangle$ **vanishes identically** if the initial and final states have incompatible symmetry with respect to $V$.

**Example: Electric Dipole Selection Rules**

For a radiative transition with dipole perturbation $V = -\boldsymbol{{d}} \cdot \boldsymbol{{E}}$ (where $\boldsymbol{{d}} = -e\boldsymbol{{r}}$):

- $\Delta \ell = \pm 1$ (orbital angular momentum changes by 1)
- $\Delta m_\ell = 0, \pm 1$ (depends on photon polarization direction)
- $\Delta S = 0$ (spin unchanged in lowest order)

**Forbidden vs. Allowed**:
- **Forbidden transition** (e.g., $2s \to 1s$, $\Delta \ell = 0$): Dipole matrix element is zero, transition rate vanishes to leading order. May proceed via higher-order processes (quadrupole, magnetic dipole) at rates suppressed by $\sim 10^{{-5}}$ or smaller.
- **Allowed transition** (e.g., $2p \to 1s$, $\Delta \ell = -1$): Non-zero dipole matrix element, normal radiative decay rate.

**Physical consequence**: Metastable states (long-lived excited states with forbidden dipole transitions) can accumulate population, used in lasers and quantum metrology.

:::

#### Interpreting Forbidden Transitions

If a transition **violates selection rules**, the dipole matrix element is zero, and the transition is **forbidden to leading order** (first-order dipole transitions). It may proceed at a much lower rate via:
- **Electric quadrupole transitions** (~10 slower)
- **Magnetic dipole transitions** (~10 slower)
- **Higher-order processes** (dramatically suppressed)

This explains why some atomic transitions have extremely long lifetimes (metastable states).

### Summary

:::{admonition} Key Results
:class: note

- **First-order transition probability**: Oscillates as $\sin^2(\omega_{fi}t/2) / \omega_{fi}^2$. Enhanced on resonance ($\omega_{fi} \approx 0$), suppressed off-resonance.
- **Sinc function**: Encodes interference. Main lobe width $\sim 2\pi/t$ sets the energy resolution.
- **Fermi's Golden Rule**: $\Gamma = (2\pi/\hbar) |V_{fi}|^2 \rho(E_i)$ with energy conservation $\delta(E_f - E_i)$.
- **Rate independence**: Once long-time limit is reached, transition rate is **constant** (no longer grows with time).
- **Coupling squared**: Doubling the perturbation strength increases the rate 4-fold.
- **Density of states**: More available final states → faster transitions.
- **Selection rules**: Symmetries determine which transitions are allowed. Violations lead to suppressed (forbidden) transitions.
- **Applications**: Radiative decay (lifetime), scattering (cross sections), absorption (materials).

:::

:::{admonition} Discussion: From Exponential Decay to Quantum Zeno Effect
:class: dropdown warning

Fermi's golden rule predicts exponential decay: $P(t) = 1 - \Gamma t$ (constant decay rate $\Gamma$). But quantum mechanics predicts $P(t) \approx 1 - (\Gamma t)^2$ for very short times (quadratic regime), not linear.

- At what timescale $t^*$ does the crossover from quadratic to linear decay occur? Use the golden rule rate $\Gamma$ as the scale.
- How does this relate to the **quantum Zeno effect**, where frequent measurements suppress decay?
- Design an experiment to measure both regimes in, e.g., an excited atom decaying by spontaneous emission.
- Does the transition depend on the form of the spectral density, or is it universal?
