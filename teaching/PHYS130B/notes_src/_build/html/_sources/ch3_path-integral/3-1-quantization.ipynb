# 3.1 Quantization

### Overview

The path to quantum mechanics begins with a fundamental question: **What is light—a particle or a wave?** For centuries, this dispute shaped physics. The answer is both. This chapter develops the **path integral** formulation of quantum mechanics, which unifies the wave and particle pictures through a single principle: quantum systems explore all possible paths, weighted by their classical action, and interference among paths determines observable outcomes.
