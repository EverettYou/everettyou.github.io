## Lecture Notes

The propagator is the quantum amplitude for a particle to travel from one spacetime point to another. For a free particle, the path integral can be evaluated exactly by summing over all trajectories, producing a Gaussian result that contains the full quantum information: spreading, interference, and the de Broglie wavelength.

### Time-Slicing: Discretizing the Path IntegralThe path integral is a functional integral—an infinite-dimensional sum over all paths $\boldsymbol{x}(t)$. To evaluate it, we **discretize** by dividing time into small intervals.

#### SetupDivide the time interval $[t_i, t_f]$ into $N$ equal slices:

$$t_i = t_0 < t_1 < t_2 < \cdots < t_N = t_f, \quad \varepsilon = \ffrac{t_f - t_i}{N}

$$ (eq-time-slicing)

The path $\boldsymbol{x}(t)$ is represented by positions $\boldsymbol{x}_n = \boldsymbol{x}(t_n)$ for $n = 0, 1, \ldots, N$.

#### Action in a Time SliceFor a path segment from $t_n$ to $t_{n+1}$ (duration $\varepsilon$), approximate the action as if the particle moves at constant velocity:

$$S_n = \int_{t_n}^{t_{n+1}} L(\boldsymbol{x}, \dot{\boldsymbol{x}}) \, dt \approx L\left(\boldsymbol{x}_n, \ffrac{\boldsymbol{x}_{n+1} - \boldsymbol{x}_n}{\varepsilon}\right) \varepsilon

$$ (eq-action-slice)

For a free particle with $L = \ffrac{m}{2}\dot{\boldsymbol{x}}^2$:

$$S_n = \ffrac{m}{2} \ffrac{(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2}{\varepsilon}

$$ (eq-free-action-slice)

The total action is

$$S = \sum_{n=0}^{N-1} S_n = \sum_{n=0}^{N-1} \ffrac{m}{2\varepsilon}(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2

$$ (eq-total-action-sliced)

#### Sliced Path IntegralThe propagator becomes

$$K(\boldsymbol{x}_N, t_f; \boldsymbol{x}_0, t_i) = \int d^3\boldsymbol{x}_1 \int d^3\boldsymbol{x}_2 \cdots \int d^3\boldsymbol{x}_{N-1} \prod_{n=0}^{N-1} \exp\left(\ffrac{im}{2\hbar\varepsilon}(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2\right)

$$ (eq-propagator-sliced)

This is a product of exponentials of quadratic forms in the positions—perfect for **Gaussian integrals**.

### Gaussian Integrals: The Key Computational Tool

#### One-Dimensional GaussianThe fundamental Gaussian integral is

$$\int_{-\infty}^{\infty} e^{-ax^2} dx = \sqrt{\ffrac{\pi}{a}}, \quad \text{Re}(a) > 0

$$ (eq-gaussian-real)

For a **complex** exponent (relevant in quantum mechanics):

$$\int_{-\infty}^{\infty} e^{iax^2} dx = \sqrt{\ffrac{\pi}{ia}} = e^{i\pi/4}\sqrt{\ffrac{\pi}{|a|}}

$$ (eq-gaussian-complex)

where we use the convention that $\sqrt{i} = e^{i\pi/4} = (1+i)/\sqrt{2}$.

#### Shifted GaussianIf the exponent is shifted:

$$\int_{-\infty}^{\infty} e^{ia(x - x_0)^2} dx = \sqrt{\ffrac{\pi}{ia}}

$$ (eq-gaussian-shifted)

The shift doesn't change the result—a key property that makes the calculation work.

#### Multi-Dimensional GaussianFor a vector $\boldsymbol{x} \in \mathbb{R}^d$ and a symmetric matrix $A$:

$$\int d^d\boldsymbol{x} \, e^{i\boldsymbol{x}^T A \boldsymbol{x}} = \ffrac{(\pi i)^{d/2}}{\sqrt{\det A}}

$$ (eq-gaussian-multidim)

Or more carefully:

$$\int d^d\boldsymbol{x} \, e^{i\boldsymbol{x}^T A \boldsymbol{x}} = \sqrt{\ffrac{\pi^d i^d}{\det A}} = \left(\ffrac{\pi}{i}\right)^{d/2} \ffrac{1}{\sqrt{\det A}}

$$ (eq-gaussian-multidim-careful)

### Computing the Free-Particle PropagatorWe evaluate the sliced path integral by integrating out positions one by one.

#### Single-Slice PropagatorThe propagator from $t_n$ to $t_{n+1}$ is

$$K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) = \int e^{\ffrac{im}{2\hbar\varepsilon}(\boldsymbol{x}' - \boldsymbol{x})^2} = \left(\ffrac{m}{2\pi i\hbar\varepsilon}\right)^{d/2} e^{\ffrac{im}{2\hbar\varepsilon}(\boldsymbol{x}' - \boldsymbol{x})^2}

$$ (eq-single-slice-prop)

The normalization factor $(m/2\pi i\hbar\varepsilon)^{d/2}$ comes from the Gaussian integral formula.

#### Two-Slice PropagatorIntegrate over the intermediate position $\boldsymbol{x}_1$:

$$K_{2\varepsilon} = \int d^d\boldsymbol{x}_1 \, K_\varepsilon(\boldsymbol{x}_2, \boldsymbol{x}_1; \varepsilon) K_\varepsilon(\boldsymbol{x}_1, \boldsymbol{x}_0; \varepsilon)

$$$$= \int d^d\boldsymbol{x}_1 \left(\ffrac{m}{2\pi i\hbar\varepsilon}\right)^d \exp\left(\ffrac{im}{2\hbar\varepsilon}[(\boldsymbol{x}_2 - \boldsymbol{x}_1)^2 + (\boldsymbol{x}_1 - \boldsymbol{x}_0)^2]\right)

$$:::{admonition} Derivation: Two-Slice Gaussian Integration
:class: dropdown note

Expand the exponent:

$$(\boldsymbol{x}_2 - \boldsymbol{x}_1)^2 + (\boldsymbol{x}_1 - \boldsymbol{x}_0)^2 = 2\boldsymbol{x}_1^2 - 2(\boldsymbol{x}_2 + \boldsymbol{x}_0)\boldsymbol{x}_1 + (\boldsymbol{x}_2^2 + \boldsymbol{x}_0^2)

$$

Completing the square:

$$= 2\left(\boldsymbol{x}_1 - frac{\boldsymbol{x}_2 + \boldsymbol{x}_0}{2}ight)^2 + (\boldsymbol{x}_2^2 + \boldsymbol{x}_0^2) - frac{(\boldsymbol{x}_2 + \boldsymbol{x}_0)^2}{2}

$$

$$= 2\left(\boldsymbol{x}_1 - frac{\boldsymbol{x}_2 + \boldsymbol{x}_0}{2}ight)^2 + frac{(\boldsymbol{x}_2 - \boldsymbol{x}_0)^2}{2}

$$

:::

The integral over $\boldsymbol{x}_1$ (a shifted Gaussian) gives:

$$\int d^d\boldsymbol{x}_1 \exp\left(\ffrac{im}{\hbar\varepsilon}\left(\boldsymbol{x}_1 - \ffrac{\boldsymbol{x}_2 + \boldsymbol{x}_0}{2}\right)^2\right) = \left(\ffrac{\pi\hbar\varepsilon}{im}\right)^{d/2}

$$Combining with the remaining exponent and normalizations:

$$K_{2\varepsilon}(\boldsymbol{x}_2, \boldsymbol{x}_0; 2\varepsilon) = \left(\ffrac{m}{2\pi i\hbar(2\varepsilon)}\right)^{d/2} \exp\left(\ffrac{im}{2\hbar(2\varepsilon)}(\boldsymbol{x}_2 - \boldsymbol{x}_0)^2\right)

$$**Pattern**: The time increases additively ($\varepsilon + \varepsilon = 2\varepsilon$), and the distance appears in the exponent with the total time.

### The Exact Free-Particle PropagatorBy induction (or by taking the continuum limit $N \to \infty$), the exact propagator for a free particle is:

$$K_\text{free}(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \left(\ffrac{m}{2\pi i\hbar T}\right)^{d/2} \exp\left(\ffrac{im}{2\hbar T}(\boldsymbol{x}_f - \boldsymbol{x}_i)^2\right)

$$ (eq-free-propagator-exact)

where $T = t_f - t_i$ is the total time interval and $d$ is the spatial dimension.

#### Physical Interpretation**Prefactor** $(m/2\pi i\hbar T)^{d/2}$:- Scales as $T^{-d/2}$ as time increases.- For $d=1$: $K \propto 1/\sqrt{T}$ (wavepacket spreads as $\sqrt{T}$).- For $d=3$: $K \propto 1/T^{3/2}$ (faster spreading in higher dimensions).**Exponent** $\ffrac{im}{2\hbar T}(\boldsymbol{x}_f - \boldsymbol{x}_i)^2$:- Equals the **classical action** for a particle traveling at constant velocity from $\boldsymbol{x}_i$ to $\boldsymbol{x}_f$:

$$S_\text{cl} = \ffrac{m v^2}{2} T = \ffrac{m}{2T}(\boldsymbol{x}_f - \boldsymbol{x}_i)^2

$$- Contains the phase that oscillates according to the classical action.**Probability density** $|K|^2$:

$$|K|^2 = \ffrac{m^d}{(2\pi\hbar T)^d} \propto T^{-d}

$$- For $d=1$: $|K|^2 \propto 1/T$ (spreading like diffusion).- For $d=3$: $|K|^2 \propto 1/T^3$ (very rapid dilution).

### Relation to the Schrödinger EquationThe free-particle propagator satisfies the **Schrödinger equation**:

$$i\hbar \ffrac{\partial K}{\partial t_f} = -\ffrac{\hbar^2}{2m} \nabla_{\boldsymbol{x}_f}^2 K

$$ (eq-schrodinger-propagator)

with **initial condition**

$$K(\boldsymbol{x}_f, t_i; \boldsymbol{x}_i, t_i) = \delta^{(d)}(\boldsymbol{x}_f - \boldsymbol{x}_i)

$$ (eq-initial-condition)

**Verification**: Taking derivatives of $K_\text{free}$ directly confirms these properties. This connection shows that the path integral is equivalent to solving the Schrödinger equation.

### Wavepacket Spreading

#### Initial Gaussian WavepacketConsider an initial wavefunction that is a **Gaussian wavepacket** (minimum uncertainty state):

$$\psi_0(\boldsymbol{x}) = \left(\ffrac{1}{\pi \hat{\sigma}_0^2}\right)^{d/4} \exp\left(-\ffrac{\boldsymbol{x}^2}{2\hat{\sigma}_0^2} + i\boldsymbol{k}_0 \cdot \boldsymbol{x}\right)

$$ (eq-initial-gaussian)

where $\hat{\sigma}_0$ is the initial width and $\boldsymbol{k}_0 = \boldsymbol{p}_0/\hbar$ is the initial momentum.

#### Time-Evolved WavepacketUsing the free-particle propagator:

$$\psi(\boldsymbol{x}, t) = \int K_\text{free}(\boldsymbol{x}, t; \boldsymbol{x}_0, 0) \, \psi_0(\boldsymbol{x}_0) \, d^d\boldsymbol{x}_0

$$The integral of two Gaussians yields another Gaussian (up to a phase). The result is:

$$\psi(\boldsymbol{x}, t) = \text{normalization} \times \exp\left(-\ffrac{(\boldsymbol{x} - \boldsymbol{x}_\text{cm}(t))^2}{2\sigma(t)^2} + i\Phi(\boldsymbol{x}, t)\right)

$$ (eq-evolved-gaussian)

where:
- **Center of mass**: $\boldsymbol{x}_\text{cm}(t) = \boldsymbol{x}_0 + (\boldsymbol{p}_0/m)t$ (classical trajectory).
- **Width**: $\sigma(t) = \hat{\sigma}_0\sqrt{1 + (\hbar t/(2m\hat{\sigma}_0^2))^2}$ (spreading).
- **Diffusion time scale**: $\tau = 2m\hat{\sigma}_0^2/\hbar$ (time for width to double).

:::{admonition} Quantum spreading vs. classical:class: noteA **classical** wavepacket has fixed width—the center moves but the shape doesn't change. A **quantum** wavepacket spreads because different momentum components travel at different velocities. For a macroscopic object ($m \sim 1$ kg, $\hat{\sigma}_0 \sim 1$ mm), the spreading time is $\tau \sim 10^{40}$ seconds—unobservable. For an electron ($m \sim 10^{-30}$ kg, $\hat{\sigma}_0 \sim 10^{-10}$ m), $\tau \sim 10^{-16}$ s—very observable. This is why microscopic systems show quantum behavior.
:::
:::{admonition} Classical-quantum correspondence:class: seealsoThe transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the correspondence principle.
:::
:::{admonition} Action and phase:class: noteIn quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::
### Summary
- **Time-slicing**: Discretize time into intervals; paths are represented by positions at each slice.
- **Gaussian integrals**: Evaluate the quadratic exponents using standard formulas; this is the computational workhorse.
- **Free-particle propagator**: Exact result is $(m/2\pi i\hbar T)^{d/2} \exp(im(\boldsymbol{x}_f - \boldsymbol{x}_i)^2/2\hbar T)$.
- **Wavepacket spreading**: The amplitude decays as $T^{-d/2}$; the width grows as $\sqrt{1 + (\hbar t/2m\hat{\sigma}_0^2)^2}$.
- **Schrödinger equivalence**: The propagator satisfies the Schrödinger equation with a delta-function initial condition.
- **Physical picture**: Quantum mechanics governs the evolution of probability amplitudes; classical action determines the phase.
