## Lecture Notes

:::{admonition} Definition
:class: important

The **propagator** $K(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i)$ is the quantum amplitude for a particle to travel from position $\boldsymbol{x}_i$ at time $t_i$ to position $\boldsymbol{x}_f$ at time $t_f$. For a free particle, the path integral can be evaluated exactly by summing over all trajectories, producing a Gaussian result that contains the full quantum information: spreading, interference, and the de Broglie wavelength.
:::

:::{admonition} Discussion: Wave Packet Spreading
:class: dropdown tip

Free-particle propagator evolves an initial delta function $\psi(\boldsymbol{x}, 0) = \delta(\boldsymbol{x} - \boldsymbol{x}_i)$ into a spreading Gaussian. The width grows as $\Delta x(t) \sim \sqrt{\hbar t / m}$. This spreading is fundamental:

- If you precisely localize a particle to within $\Delta x_0$ using position measurement, its momentum uncertainty grows to $\Delta p \sim \hbar/\Delta x_0$ (uncertainty principle). After time $t$, the wave packet width should be $\Delta x(t) \sim \sqrt{\Delta p^2 t^2 / m^2 + \Delta x_0^2}$. Can you reconcile this with the free-particle propagator formula?
- Quantum spreading ($\Delta x \propto \sqrt{t}$) is slower than classical diffusion ($\Delta x \propto \sqrt{t}$, same scaling!) yet faster than ballistic expansion. Why does the diffusion-like $\sqrt{t}$ behavior emerge quantum mechanically without any dissipation?
- Consider a macroscopic baseball: $m = 0.1$ kg, initially localized to $\Delta x_0 = 1$ mm. How long does it take for quantum spreading to double the size? Does this explain why we don't observe quantum superposition in macroscopic objects?
- If you measure the particle's momentum after initial position measurement, you prepare a momentum eigenstate with sharp $p$. Its position uncertainty then grows as $\Delta x(t) = \Delta x_0 + p t / m$ (no spreading factor!). Is this consistent with the propagator?
:::
### Time-Slicing: Discretizing the Path Integral

The path integral is a functional integral—an infinite-dimensional sum over all paths $\boldsymbol{x}(t)$. To evaluate it, we **discretize** by dividing time into small intervals.

#### Setup

Divide the time interval $[t_i, t_f]$ into $N$ equal slices:

$$t_i = t_0 < t_1 < t_2 < \cdots < t_N = t_f, \quad \varepsilon = \frac{t_f - t_i}{N}$$


The path $\boldsymbol{x}(t)$ is represented by positions $\boldsymbol{x}_n = \boldsymbol{x}(t_n)$ for $n = 0, 1, \ldots, N$.

#### Action in a Time Slice

For a path segment from $t_n$ to $t_{n+1}$ (duration $\varepsilon$), approximate the action as if the particle moves at constant velocity:

$$S_n = \int_{t_n}^{t_{n+1}} L(\boldsymbol{x}, \dot{\boldsymbol{x}}) \, \mathrm{d}t \approx L\left(\boldsymbol{x}_n, \frac{\boldsymbol{x}_{n+1} - \boldsymbol{x}_n}{\varepsilon}\right) \varepsilon$$


For a free particle with $L = \frac{m}{2}\dot{\boldsymbol{x}}^2$:

$$S_n = \frac{m}{2} \frac{(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2}{\varepsilon}$$


The total action is

$$S = \sum_{n=0}^{N-1} S_n = \sum_{n=0}^{N-1} \frac{m}{2\varepsilon}(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2$$


#### Sliced Path Integral

The propagator becomes

$$K(\boldsymbol{x}_N, t_f; \boldsymbol{x}_0, t_i) = \int d^3\boldsymbol{x}_1 \int d^3\boldsymbol{x}_2 \cdots \int d^3\boldsymbol{x}_{N-1} \prod_{n=0}^{N-1} \exp\left(\frac{\mathrm{i}m}{2\hbar\varepsilon}(\boldsymbol{x}_{n+1} - \boldsymbol{x}_n)^2\right)$$


This is a product of exponentials of quadratic forms in the positions—perfect for **Gaussian integrals**.

:::{admonition} Important Result
:class: important

The fundamental Gaussian integral is

$$\int_{-\infty}^{\infty} \mathrm{e}^{-ax^2} dx = \sqrt{\frac{\pi}{a}}, \quad \text{Re}(a) > 0$$


For **complex** exponents (relevant in quantum mechanics):

$$\int_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i}ax^2} dx = \sqrt{\frac{\pi}{ia}} = \mathrm{e}^{\mathrm{i}\pi/4}\sqrt{\frac{\pi}{|a|}}$$


where $\sqrt{i} = \mathrm{e}^{\mathrm{i}\pi/4} = (1+i)/\sqrt{2}$.
:::

**Shifted Gaussian**: If the exponent is shifted:

$$\int_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i}a(x - x_0)^2} dx = \sqrt{\frac{\pi}{ia}}$$


The shift doesn't change the result—a key property that makes the calculation work.

**Multi-Dimensional Gaussian**: For a vector $\boldsymbol{x} \in \mathbb{R}^d$ and a symmetric matrix $A$:

$$\int d^d\boldsymbol{x} \, \mathrm{e}^{\mathrm{i}\boldsymbol{x}^T A \boldsymbol{x}} = \sqrt{\frac{\pi^d i^d}{\det A}} = \left(\frac{\pi}{\mathrm{i}}\right)^{d/2} \frac{1}{\sqrt{\det A}}$$


### Computing the Free-Particle Propagator

We evaluate the sliced path integral by integrating out positions one by one.

#### Single-Slice Propagator

The propagator from $t_n$ to $t_{n+1}$ is

$$K_\varepsilon(\boldsymbol{x}', \boldsymbol{x}; \varepsilon) = \int \mathrm{e}^{\frac{\mathrm{i}m}{2\hbar\varepsilon}(\boldsymbol{x}' - \boldsymbol{x})^2} = \left(\frac{m}{2\pi \mathrm{i}\hbar\varepsilon}\right)^{d/2} \mathrm{e}^{\frac{\mathrm{i}m}{2\hbar\varepsilon}(\boldsymbol{x}' - \boldsymbol{x})^2}$$


The normalization factor $(m/2\pi \mathrm{i}\hbar\varepsilon)^{d/2}$ comes from the Gaussian integral formula.

#### Two-Slice Propagator

Integrate over the intermediate position $\boldsymbol{x}_1$:

$$K_{2\varepsilon} = \int d^d\boldsymbol{x}_1 \, K_\varepsilon(\boldsymbol{x}_2, \boldsymbol{x}_1; \varepsilon) K_\varepsilon(\boldsymbol{x}_1, \boldsymbol{x}_0; \varepsilon)$$


$$= \int d^d\boldsymbol{x}_1 \left(\frac{m}{2\pi \mathrm{i}\hbar\varepsilon}\right)^d \exp\left(\frac{\mathrm{i}m}{2\hbar\varepsilon}[(\boldsymbol{x}_2 - \boldsymbol{x}_1)^2 + (\boldsymbol{x}_1 - \boldsymbol{x}_0)^2]\right)$$


:::{admonition} Derivation: Two-Slice Gaussian Integration
:class: dropdown information

Expand the exponent:

$$( \boldsymbol{x}_2 - \boldsymbol{x}_1)^2 + (\boldsymbol{x}_1 - \boldsymbol{x}_0)^2 = 2\boldsymbol{x}_1^2 - 2(\boldsymbol{x}_2 + \boldsymbol{x}_0) \cdot \boldsymbol{x}_1 + (\boldsymbol{x}_2^2 + \boldsymbol{x}_0^2)$$


Complete the square:

$$= 2\left(\boldsymbol{x}_1 - \frac{\boldsymbol{x}_2 + \boldsymbol{x}_0}{2}\right)^2 + (\boldsymbol{x}_2^2 + \boldsymbol{x}_0^2) - \frac{(\boldsymbol{x}_2 + \boldsymbol{x}_0)^2}{2}$$


$$= 2\left(\boldsymbol{x}_1 - \frac{\boldsymbol{x}_2 + \boldsymbol{x}_0}{2}\right)^2 + \frac{(\boldsymbol{x}_2 - \boldsymbol{x}_0)^2}{2}$$


The integral over $\boldsymbol{x}_1$ (a shifted Gaussian) gives:

$$\int d^d\boldsymbol{x}_1 \exp\left(\frac{\mathrm{i}m}{\hbar\varepsilon}\left(\boldsymbol{x}_1 - \frac{\boldsymbol{x}_2 + \boldsymbol{x}_0}{2}\right)^2\right) = \left(\frac{\pi\hbar\varepsilon}{im}\right)^{d/2}$$


Combining with the remaining exponent and normalizations:

$$K_{2\varepsilon}(\boldsymbol{x}_2, \boldsymbol{x}_0; 2\varepsilon) = \left(\frac{m}{2\pi \mathrm{i}\hbar(2\varepsilon)}\right)^{d/2} \exp\left(\frac{\mathrm{i}m}{2\hbar(2\varepsilon)}(\boldsymbol{x}_2 - \boldsymbol{x}_0)^2\right)$$


**Key observation**: The time increases additively ($\varepsilon + \varepsilon = 2\varepsilon$), and the distance appears in the exponent with the total time. This pattern generalizes to $N$ slices.
:::

### The Exact Free-Particle Propagator

By induction (or by taking the continuum limit $N \to \infty$), the exact propagator for a free particle is:

:::{admonition} The Free-Particle Propagator
:class: important

$$K_\text{free}(\boldsymbol{x}_f, t_f; \boldsymbol{x}_i, t_i) = \left(\frac{m}{2\pi \mathrm{i}\hbar T}\right)^{d/2} \exp\left(\frac{\mathrm{i}m}{2\hbar T}(\boldsymbol{x}_f - \boldsymbol{x}_i)^2\right)$$


where $T = t_f - t_i$ is the total time interval and $d$ is the spatial dimension.
:::

#### Physical Interpretation

**Prefactor** $(m/2\pi \mathrm{i}\hbar T)^{d/2}$:
- Scales as $T^{-d/2}$ as time increases
- For $d=1$: $K \propto 1/\sqrt{T}$ (wavepacket spreads as $\sqrt{T}$)
- For $d=3$: $K \propto 1/T^{3/2}$ (faster spreading in higher dimensions)

**Exponent** $\frac{\mathrm{i}m}{2\hbar T}(\boldsymbol{x}_f - \boldsymbol{x}_i)^2$:
- Equals the **classical action** for a particle traveling at constant velocity from $\boldsymbol{x}_i$ to $\boldsymbol{x}_f$:

$$S_\text{cl} = \frac{m v^2}{2} T = \frac{m}{2T}(\boldsymbol{x}_f - \boldsymbol{x}_i)^2$$


- Contains the phase that oscillates according to the classical action

**Probability density** $|K|^2$:

$$|K|^2 = \frac{m^d}{(2\pi\hbar T)^d} \propto T^{-d}$$


- For $d=1$: $|K|^2 \propto 1/T$ (spreading like diffusion)
- For $d=3$: $|K|^2 \propto 1/T^3$ (very rapid dilution)

:::{admonition} The Propagator Satisfies the Schrödinger Equation
:class: important

$$\mathrm{i}\hbar \frac{\partial K}{\partial t_f} = -\frac{\hbar^2}{2m} \nabla_{\boldsymbol{x}_f}^2 K$$


with **initial condition**

$$K(\boldsymbol{x}_f, t_i; \boldsymbol{x}_i, t_i) = \delta^{(d)}(\boldsymbol{x}_f - \boldsymbol{x}_i)$$


This connection shows that the path integral is equivalent to solving the Schrödinger equation. Verification by direct differentiation of $K_\text{free}$ is straightforward but algebraically involved.
:::

### Wavepacket Spreading

#### Initial Gaussian Wavepacket

Consider an initial wavefunction that is a **Gaussian wavepacket** (minimum uncertainty state):

$$\psi_0(\boldsymbol{x}) = \left(\frac{1}{\pi \sigma_0^2}\right)^{d/4} \exp\left(-\frac{\boldsymbol{x}^2}{2\sigma_0^2} + \mathrm{i}\boldsymbol{k}_0 \cdot \boldsymbol{x}\right)$$


where $\sigma_0$ is the initial width and $\boldsymbol{k}_0 = \boldsymbol{p}_0/\hbar$ is the initial momentum.

#### Time-Evolved Wavepacket

Using the free-particle propagator:

$$\psi(\boldsymbol{x}, t) = \int K_\text{free}(\boldsymbol{x}, t; \boldsymbol{x}_0, 0) \, \psi_0(\boldsymbol{x}_0) \, d^d\boldsymbol{x}_0$$


The integral of two Gaussians yields another Gaussian (up to a phase). The result is:

:::{admonition} Time Evolution of a Gaussian Wavepacket
:class: important

$$\psi(\boldsymbol{x}, t) = \text{normalization} \times \exp\left(-\frac{(\boldsymbol{x} - \boldsymbol{x}_\text{cm}(t))^2}{2\sigma(t)^2} + i\Phi(\boldsymbol{x}, t)\right)$$


where:
- **Center of mass**: $\boldsymbol{x}_\text{cm}(t) = \boldsymbol{x}_0 + (\boldsymbol{p}_0/m)t$ (classical trajectory)
- **Width**: $\sigma(t) = \sigma_0\sqrt{1 + (\hbar t/(2m\sigma_0^2))^2}$ (spreading)
- **Spreading time scale**: $\tau = 2m\sigma_0^2/\hbar$ (time for width to double)
:::

:::{admonition} Note: Quantum Spreading vs. Classical
:class: note

A **classical** wavepacket has fixed width—the center moves but the shape doesn't change. A **quantum** wavepacket spreads because different momentum components travel at different velocities. 

For a macroscopic object ($m \sim 1$ kg, $\sigma_0 \sim 1$ mm), the spreading time is $\tau \sim 10^{40}$ seconds—unobservable. For an electron ($m \sim 10^{-30}$ kg, $\sigma_0 \sim 10^{-10}$ m), $\tau \sim 10^{-16}$ s—very observable. This is why microscopic systems show quantum behavior.
:::

:::{admonition} Note: Classical-Quantum Correspondence
:class: note

The transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the **correspondence principle**.

In quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::

### Summary

- **Time-slicing**: Discretize time into intervals; paths are represented by positions at each slice
- **Gaussian integrals**: Evaluate the quadratic exponents using standard formulas; this is the computational workhorse
- **Free-particle propagator**: Exact result is $(m/2\pi \mathrm{i}\hbar T)^{d/2} \exp(\mathrm{i}m(\boldsymbol{x}_f - \boldsymbol{x}_i)^2/2\hbar T)$
- **Wavepacket spreading**: The amplitude decays as $T^{-d/2}$; the width grows as $\sqrt{1 + (\hbar t/2m\sigma_0^2)^2}$
- **Schrödinger equivalence**: The propagator satisfies the Schrödinger equation with a delta-function initial condition
- **Physical picture**: Quantum mechanics governs the evolution of probability amplitudes; classical action determines the phase
