## Lecture Notes

De Broglie's insight unifies matter and light by assigning waves to particles, with wavelength inversely proportional to momentum. This transforms the Schrödinger equation into a wave equation for matter, where quantization arises naturally from the requirement that the wavefunction be single-valued around closed orbits—the Bohr-Sommerfeld condition.



:::{admonition} Discussion: Wave-Geometric Duality in Quantum and Optics
:class: dropdown warning

Schrödinger equation has the same form as the wave equation (with $\hbar\omega$ instead of $c^2 k^2$), and action enters the phase just like optical phase. This suggests deep structural kinship:

- In geometric optics, rays follow paths determined by Fermat's principle (stationary optical path). In classical mechanics, particles follow Euler-Lagrange equations (stationary action). Are these two instances of the same principle?
- When light passes through a potential (varying refractive index), the ray bends via Snell's law. When a quantum particle encounters a potential, its wavelength changes via $p = h/\lambda$. Can these be unified in a single framework?
- Consider a wave packet of light entering a dispersive medium (refractive index depends on frequency). The wave packet spreads and disperses. Does a quantum wavepacket in a potential behave similarly? Can you draw parallels between optical and quantum dispersion?
- The eikonal approximation (geometric limit) applies when wavelength is small compared to obstacle size. For quantum mechanics, $\lambda = h/p$ becomes small when momentum is large. Does this explain why macroscopic objects follow classical trajectories?
:::
### The Matter Wave: De Broglie Relation

At the heart of quantum mechanics lies a simple but profound observation:

:::{admonition} De Broglie Hypothesis (1924)
:class: important

Particles possess wave properties with wavelength

$$


\lambda = \frac{h}{p} = \frac{2\pi\hbar}{p}

$$ (eq-de-broglie)

and frequency

$$


f = \frac{E}{h}, \quad \text{or} \quad \omega = \frac{E}{\hbar}

$$ (eq-de-broglie-frequency)

where $p$ is the particle's momentum, $E$ is its energy, and $h = 2\pi\hbar$ is Planck's constant.
:::

This is the **inverse** of Planck-Einstein relations. Whereas light (a wave) carries momentum $p = \hbar k$ and energy $E = \hbar\omega$, matter (a particle) exhibits waves with those same relations.

**Experimental confirmation**: Davisson and Germer (1927) passed electrons through a crystal and observed diffraction patterns—proof that electrons behave as waves. The de Broglie wavelength was confirmed to atomic precision.

#### Wave Vector and Momentum

For a plane wave $\mathrm{e}^{\mathrm{i}(\boldsymbol{k}\cdot\boldsymbol{r} - \omega t)}$:

$$


\text{Spatial oscillation: } \lambda = \frac{2\pi}{k} \quad \Rightarrow \quad p = \hbar k

$$ (eq-momentum-k)

$$


\text{Temporal oscillation: } T = \frac{2\pi}{\omega} \quad \Rightarrow \quad E = \hbar\omega

$$ (eq-energy-omega)

The **wave vector** $\boldsymbol{k}$ is related to momentum as $\boldsymbol{p} = \hbar\boldsymbol{k}$. A particle with definite momentum $p$ has definite wavelength $\lambda = 2\pi\hbar/p$.

### Hamilton-Jacobi Meets Eikonal: From Action to Wave

Classical mechanics is based on the **principle of stationary action**:

$$


\delta S[\boldsymbol{x}] = 0, \quad S = \int L(\boldsymbol{x}, \dot{\boldsymbol{x}}) \, \mathrm{d}t

$$ (eq-stationary-action)

For a non-relativistic particle with Lagrangian $L = \frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})$, the Hamiltonian is $H = \frac{\boldsymbol{p}^2}{2m} + V(\boldsymbol{x})$, and the **Hamilton-Jacobi equation** for the action $S(\boldsymbol{x}, t)$ is

$$


\frac{\partial S}{\partial t} + \frac{1}{2m}(\nabla S)^2 + V(\boldsymbol{x}) = 0

$$ (eq-hamilton-jacobi)

where $\boldsymbol{p} = \nabla S$ (by the definition of conjugate momentum).

#### The Quantum Replacement: $S \to \hbar \Phi$

In quantum mechanics, we promote the action to a **phase function**:

$$


S(\boldsymbol{x}, t) \longrightarrow \hbar \, \Phi(\boldsymbol{x}, t)

$$ (eq-action-to-phase)

And the **wavefunction** is written as

$$


\psi(\boldsymbol{x}, t) = A(\boldsymbol{x}, t) \, \mathrm{e}^{\mathrm{i}\Phi(\boldsymbol{x}, t)}

$$ (eq-wavefunction-amplitude-phase)

where $A$ is the amplitude and $\Phi$ is the phase (in radians, dimensionless). The physical phase carried by the wavefunction is $\Phi(\boldsymbol{x}, t)$.

#### Connecting to the Schrödinger Equation

$$ (eq-schrodinger-td)

:class: note

The decomposition $\psi = A \mathrm{e}^{\mathrm{i}\Phi}$ reveals the quantum mechanical content:

- **Phase $\Phi$**: Obeys the **Hamilton-Jacobi equation**—evolves classically.

- **Amplitude $A$**: Determines **probability density** $|\psi|^2 = A^2$—spreads and disperses due to quantum effects.

In the **classical limit** where quantum spreading is negligible, both amplitude and phase become sharply localized to the classical trajectory.

### The Wavefunction as Probability Amplitude

The **Born rule** postulates that the probability density for finding a particle at position $\boldsymbol{x}$ at time $t$ is

$$


P(\boldsymbol{x}, t) = |\psi(\boldsymbol{x}, t)|^2

$$ (eq-born-rule)

The wavefunction is a **probability amplitude**—its modulus squared gives probability density.

#### Plane Wave: Free Particle

For a free particle with momentum $p$ and energy $E = p^2/(2m)$, the wavefunction is

$$


\psi(\boldsymbol{x}, t) = N \, \mathrm{e}^{\mathrm{i}(\boldsymbol{k}\cdot\boldsymbol{x} - \omega t)}

$$ (eq-plane-wave)

where $k = p/\hbar$ and $\omega = E/\hbar$. The phase is $\Phi = \boldsymbol{k}\cdot\boldsymbol{x} - \omega t$.

**Probability density**: $|\psi|^2 = |N|^2$ is **uniform in space**—a definite momentum state is completely delocalized! This is the **uncertainty principle**: definite momentum means completely uncertain position.

#### Wavepacket: Superposition of Momentum Eigenstates

To localize a particle, we superpose plane waves with nearby momenta:

$$


\psi(\boldsymbol{x}, t) = \int d^3\boldsymbol{k} \, \phi(\boldsymbol{k}) \, \mathrm{e}^{\mathrm{i}(\boldsymbol{k}\cdot\boldsymbol{x} - \omega_k t)}

$$ (eq-wavepacket)

where $\phi(\boldsymbol{k})$ is the **momentum distribution** and $\omega_k = \hbar k^2/(2m)$ for a free particle. The width in position space is inversely related to the width in momentum space: $\Delta x \Delta p \gtrsim \hbar/2$ (Heisenberg uncertainty).

### Quantization by Interference: Bohr's Model Generalized

In the **Bohr model** of hydrogen, the electron orbits the nucleus with quantized angular momenta. The condition is that an integer number of de Broglie wavelengths fit around the orbit:

$$


2\pi r = n \lambda = n \frac{h}{p}

$$ (eq-bohr-quantization)

This leads to quantized orbits and quantized energy levels.

#### General Quantization Rule

For any bound state in a potential $V(\boldsymbol{x})$, **quantization arises from the requirement that the wavefunction be single-valued and well-behaved**. The wavefunction must satisfy:

1. **Wave equation**: It obeys the Schrödinger equation.
2. **Boundary conditions**: It vanishes at infinity (or is periodic for periodic potentials).
3. **Single-valuedness**: As we traverse a closed path in configuration space, the phase must increase by a multiple of $2\pi$.

In the **semiclassical limit**, this translates to the **Bohr-Sommerfeld quantization rule**:

$$


\oint \boldsymbol{p} \cdot d\boldsymbol{x} = 2\pi\hbar(n + \delta), \quad n = 0, 1, 2, \ldots

$$ (eq-bohr-sommerfeld)

where the integral is around a closed classical orbit and $\delta$ is a phase correction (typically $1/2$).

:::{admonition} Why Quantization Works
:class: important

Quantization emerges because the phase accumulated around a closed orbit must be a multiple of $2\pi$ for the wavefunction to be single-valued. Since the phase is $\int \boldsymbol{p} \cdot d\boldsymbol{x} / \hbar$, we get

$$


\frac{1}{\hbar} \oint \boldsymbol{p} \cdot d\boldsymbol{x} = 2\pi(n + \delta)

$$


This condition selects discrete energies—the bound-state energies—without assuming any additional quantum postulates beyond the Schrödinger equation and single-valuedness of the wavefunction.
:::

### The Schrödinger Equation as the Wave Equation for Matter

Just as Maxwell's equations govern light waves, the **Schrödinger equation** governs matter waves:

$$


\mathrm{i}\hbar \frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m} \nabla^2 \psi + V(\boldsymbol{x}) \psi

$$ (eq-schrodinger-standard)

**Structure**:

- **Left side**: time derivative of $\psi$ (wave propagation in time).

- **First term on right**: kinetic energy (from the Laplacian—spatial curvature of the wavefunction).

- **Second term on right**: potential energy.

This is the **quantum mechanical equation of motion**. It has the same role as Newton's $F = ma$ in classical mechanics or Maxwell's equations in electrodynamics.

#### Energy Eigenstates

For stationary states with energy $E$, we separate variables: $\psi(\boldsymbol{x}, t) = \phi(\boldsymbol{x}) \mathrm{e}^{-\mathrm{i}Et/\hbar}$. Substituting into the Schrödinger equation gives the **time-independent Schrödinger equation**:

$$


E \phi(\boldsymbol{x}) = -\frac{\hbar^2}{2m} \nabla^2 \phi + V(\boldsymbol{x}) \phi

$$ (eq-schrodinger-time-independent)

This is a **Sturm-Liouville eigenvalue problem**. Only discrete energies $E_n$ (eigenvalues) produce normalizable solutions (eigenfunctions). These are the **energy levels**.

### The Path Integral as Quantum Optics

Huygens' principle for light says: **the amplitude at a point is the superposition of wavelets from all intermediate points on the wavefront, each weighted by its phase.**

For matter, the **path integral** is the analogous principle: **the quantum amplitude from one spacetime point to another is the superposition of amplitudes from all possible paths connecting them, each weighted by the phase $\mathrm{e}^{\mathrm{i}S/\hbar}$.**

$$


\psi(\boldsymbol{x}_f, t_f) = \int_{\boldsymbol{x}_i \to \boldsymbol{x}_f} \mathrm{e}^{\mathrm{i}S[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}] \, \psi(\boldsymbol{x}_i, t_i)

$$ (eq-path-integral-matter)

where $S$ is the classical action and the integral is a **functional integral** over all paths.

**Connection to optics**:

- **Optical path length** $L = \int n \, \mathrm{d}s$ → **Classical action** $S = \int L \, \mathrm{d}t$

- **Refractive index** $n$ → **Lagrangian** $L = \frac{m}{2}\dot{\boldsymbol{x}}^2 - V$

- **Light wavelength** $\lambda$ → **de Broglie wavelength** $\lambda = h/p$

- **Huygens' principle** → **Path integral**

### Summary

- **De Broglie relation**: Particles have wavelength $\lambda = h/p$ and frequency $f = E/h$—unifying particle and wave pictures.

- **Hamilton-Jacobi to Schrödinger**: The phase of the wavefunction obeys the classical Hamilton-Jacobi equation; the amplitude evolves via a continuity equation.

- **Probability amplitude**: The wavefunction $\psi$ is a complex amplitude; $|\psi|^2$ is probability density.

- **Quantization by interference**: Single-valuedness of $\psi$ around closed orbits enforces the Bohr-Sommerfeld rule.

- **Schrödinger equation**: The quantum equation of motion, analogous to Maxwell's equations for light or Newton's laws for particles.

- **Path integral principle**: Quantum amplitude is a sum over all paths weighted by $\mathrm{e}^{\mathrm{i}S/\hbar}$—Huygens' principle for matter waves.
