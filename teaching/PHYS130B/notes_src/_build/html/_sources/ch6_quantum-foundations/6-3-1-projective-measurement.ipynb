## Lecture Notes

Measurement in quantum mechanics is not a passive observation—it is a projective act that forces the system to choose between distinct alternatives. Each possible outcome corresponds to a subspace of Hilbert space, and the measurement probability is determined by how much of the wavefunction points along that subspace. This section develops the mathematical machinery that explains why measurement causes such profound changes.

### Spectral Decomposition and Hermitian Operators

For a Hermitian operator $A$ (observable), the **spectral theorem** states:

$$

A = \sum_i a_i P_i

$$ (eq-projmeas-spectral)

where:
- $a_i$ = eigenvalues (real, since $A$ is Hermitian).
- $P_i$ = projector onto eigenspace of $a_i$.
- $P_i$ are **orthogonal**: $P_i P_j = \delta_{ij} P_i$.
- $P_i$ are **complete**: $\sum_i P_i = I$ (sum to identity).
- $P_i^2 = P_i$ (idempotent).

**Degeneracy**: If eigenvalue $a_i$ has multiplicity $> 1$, projector $P_i$ projects onto a multi-dimensional subspace.

:::{admonition} Example: Qubit Spin Observable
:class: example


For $\hat{\sigma}_z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ (Pauli-z operator):

- Eigenvalues: $\lambda_1 = +1, \lambda_2 = -1$.
- Eigenvectors: $|\uparrow\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, |\downarrow\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.
- Projectors:

$$

P_+ = |\uparrow\rangle\langle\uparrow| = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}, \quad P_- = |\downarrow\rangle\langle\downarrow| = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix}

$$

- Verification: $P_+ + P_- = I$ ✓, $P_+^2 = P_+$ ✓, $P_+ P_- = 0$ ✓.
- Spectral form: $\hat{\sigma}_z = (+1) P_+ + (-1) P_-$.

:::
### The Measurement Postulate

Recall the **measurement postulate** from §1.2.1: For a system in state $\rho$ measured for observable $A = \sum_i a_i P_i$, the outcome is eigenvalue $a_i$ with probability $p_i = \text{Tr}(P_i \rho)$, and the state collapses to $\rho' = P_i \rho P_i / \text{Tr}(P_i \rho)$.

Here we explore the mathematical structure more carefully, generalizations, and physical consequences.

**Outcome**: Measurement yields eigenvalue $a_i$ with **probability**:

$$

p_i = \text{Tr}(P_i \rho)

$$ (eq-projmeas-probability)

**Post-measurement state** (state collapse):

$$

\rho' = \frac{P_i \rho P_i}{\text{Tr}(P_i \rho)} = \frac{P_i \rho P_i}{p_i}

$$ (eq-projmeas-collapse)

The state is **projected** onto the eigenspace corresponding to the measured eigenvalue.

#### Normalization Check

$$

\text{Tr}(\rho') = \frac{\text{Tr}(P_i \rho P_i)}{p_i} = \frac{p_i}{p_i} = 1 \quad ✓

$$

since $P_i^2 = P_i$ and $P_i$ is Hermitian.

#### Expectation Value

The average outcome of repeated measurements is:

$$

\langle A \rangle = \sum_i a_i p_i = \sum_i a_i \text{Tr}(P_i \rho) = \text{Tr}(A \rho)

$$ (eq-projmeas-expectation)

This is the **Born rule** in density matrix form.

### Pure State Measurement

For a **pure state** $\rho = |\psi\rangle\langle\psi|$:

**Probability**:

$$

p_i = \text{Tr}(P_i |\psi\rangle\langle\psi|) = \langle\psi| P_i |\psi\rangle

$$ (eq-projmeas-pure-prob)

If $|\psi\rangle = \sum_j c_j |u_j^{(i)}\rangle$ where $|u_j^{(i)}\rangle$ are eigenstates of eigenvalue $a_i$ (with multiplicity $d_i$):

$$

p_i = \sum_j |c_j|^2

$$

**Post-measurement state**:

$$

|\psi'\rangle = \frac{P_i |\psi\rangle}{\sqrt{p_i}}

$$ (eq-projmeas-pure-collapse)

The state is **normalized projection** onto eigenspace $i$.

:::{admonition} Example: Measuring $\hat{\sigma}_z$ on Superposition
:class: example


Initial state: $|\psi\rangle = \frac{1}{\sqrt{2}}(|\uparrow\rangle + |\downarrow\rangle)$ (superposition).

Measure $\hat{\sigma}_z$:
- **Outcome** $+1$ (spin-up):
  - Probability: $p_+ = |\frac{1}{\sqrt{2}}|^2 = 1/2$.
  - Post-measurement: $|\psi'\rangle = |\uparrow\rangle$ (definite spin-up).

- **Outcome** $-1$ (spin-down):
  - Probability: $p_- = |\frac{1}{\sqrt{2}}|^2 = 1/2$.
  - Post-measurement: $|\psi'\rangle = |\downarrow\rangle$ (definite spin-down).

After measurement, the superposition is **destroyed**; the state is definite.

:::
### Repeated Measurements

**Theorem**: If you measure observable $A$ and obtain outcome $a_i$, then immediately measure $A$ again, you get the **same outcome** $a_i$ with **certainty** (probability 1).

**Proof**: After first measurement, state is $\rho' = P_i \rho P_i / p_i$, which is **supported** on eigenspace of $a_i$.

Second measurement probability:

$$

p_i^{(2)} = \text{Tr}(P_i \rho') = \text{Tr}\left(P_i \frac{P_i \rho P_i}{p_i}\right) = \frac{\text{Tr}(P_i P_i \rho P_i)}{p_i} = \frac{\text{Tr}(P_i \rho P_i)}{p_i} = \frac{p_i}{p_i} = 1

$$

**Consequence**: Measurement outcome is perfectly deterministic on repeated measurement; the result **doesn't change**.

### The Quantum Zeno Effect

**Setup**: System starts in state $|\psi_0\rangle$ that would evolve to $|\psi(t)\rangle$ under unitary evolution (e.g., decay).

**Frequent projective measurements**: Measure at times $0, \Delta t, 2\Delta t, \ldots$ whether the system is still in $|\psi_0\rangle$ (projector $P = |\psi_0\rangle\langle\psi_0|$).

**Result (Zeno's paradox)**: As $\Delta t \to 0$ (infinite measurement rate), the system **remains** in $|\psi_0\rangle$; evolution is **frozen**.

**Mechanism**: Each measurement collapses the state back to the measured outcome. If measurements are frequent enough, evolution is prevented before the system can deviate significantly.

Consider the effect quantitatively:

:::{admonition} Derivation: Quantum Zeno Effect
:class: dropdown note

After time $\Delta t$ without measurement, state evolves to $|\psi(\Delta t)\rangle \approx |\psi_0\rangle - i H |\psi_0\rangle \Delta t/\hbar$ (first order).

Overlap: $|\langle\psi_0|\psi(\Delta t)\rangle|^2 \approx 1 - O((\Delta t)^2)$.

After measurement: State either stays in $|\psi_0\rangle$ (probability $\approx 1 - O((\Delta t)^2)$) or exits (probability $\propto (\Delta t)^2$).

With $n = T/\Delta t$ measurements over time $T$:

$$

P(\text{remain in } |\psi_0\rangle) \approx (1 - c(\Delta t)^2)^{T/\Delta t} \to 1 \quad \text{as } \Delta t \to 0

$$

:::
### Degenerate Eigenvalues

If eigenvalue $a_i$ is **degenerate** (multiple eigenstates), the projector projects onto the entire eigenspace:

$$

P_i = \sum_{j=1}^{d_i} |u_j^{(i)}\rangle\langle u_j^{(i)}|

$$

where $d_i$ is the multiplicity (degeneracy).

**Post-measurement state**: 

$$

\rho' = \frac{P_i \rho P_i}{p_i}

$$

is supported on the entire eigenspace (possibly mixed, if initial state was a superposition of degenerate eigenstates).

:::{admonition} Example: Measuring Spin-Up/Down vs. Angular Momentum Squared
:class: example


- **$\hat{\sigma}_z$ measurement**: Non-degenerate eigenvalues (+1, -1); outcome fully specifies the state.
- **$L^2$ measurement** (orbital angular momentum squared): Degenerate eigenvalues ($l(l+1)\hbar^2$ with degeneracy $2l+1$); outcome specifies angular momentum but not magnetic quantum number.

:::
### Compatibility and Commuting Observables

Two observables $A$ and $B$ **commute** ($[A, B] = 0$) iff they have a **simultaneous eigenbasis**. In this case:
- Can measure both without disturbing the state (outcome of first doesn't affect second).
- Can assign definite values to both.

**Non-commuting observables** (e.g., $\hat{\sigma}_x$ and $\hat{\sigma}_z$):
- Measuring $A$ disturbs $B$ (collapses to eigenstate of $A$, which is superposition of $B$ eigenstates).
- Cannot assign simultaneous definite values.
- Measurement outcomes are **incompatible**.

:::{admonition} Projective measurement is a special case
:class: note
Projective (von Neumann) measurements are the textbook case where outcomes correspond to orthogonal projectors summing to identity. The more general framework (POVM) relaxes orthogonality—see §6.3.2.
:::

### Summary

- **Projective measurement**: von Neumann measurement based on spectral decomposition $A = \sum_i a_i P_i$.
- **Measurement postulate** (detailed in §1.2.1; extended here): Outcome $a_i$ with probability $p_i = \text{Tr}(P_i \rho)$; state collapses to $P_i \rho P_i / p_i$.
- **Expectation value**: $\langle A \rangle = \text{Tr}(A\rho)$ (Born rule).
- **Pure states**: $p_i = \langle\psi| P_i |\psi\rangle$; post-measurement $|\psi'\rangle = P_i|\psi\rangle / \sqrt{p_i}$.
- **Repeated measurements**: Second measurement of same observable gives same outcome with certainty (measurement result is reproducible).
- **Quantum Zeno effect**: Frequent measurements freeze the state, preventing evolution.
- **Degeneracy**: Multiple eigenstates for same eigenvalue; projector onto entire eigenspace.
- **Commuting observables**: Share eigenbasis; can be measured simultaneously without disturbance.
- **Non-commuting**: Mutually incompatible; measuring one disturbs the other.
