## Lecture Notes

When two or more quantum systems are brought together, their combined state lives in a vastly larger space—a tensor product Hilbert space. Yet something remarkable can happen: the total system cannot be decomposed into independent pieces, even if we try. This phenomenon of entanglement reveals that quantum correlations are fundamentally different from classical ones: measuring one part instantly affects the statistics for the other, no matter how far apart they are.

### Tensor Product Hilbert Space



:::{admonition} Discussion
:class: dropdown warning

The Schmidt decomposition is unique when Schmidt coefficients are non-degenerate. What happens when two Schmidt coefficients are equal? Does this affect entanglement measures?
:::


:::{admonition} Definition: Tensor Product Hilbert Space
:class: important

For two quantum systems $A$ and $B$ with Hilbert spaces $\mathcal{H}_A$ (dimension $d_A$) and $\mathcal{H}_B$ (dimension $d_B$), the **composite system** $AB$ lives in the **tensor product** space:

$$\mathcal{H}_{AB} = \mathcal{H}_A \otimes \mathcal{H}_B

$$ (eq-composite-tensor)

with dimension $\dim(\mathcal{H}_{AB}) = d_A \cdot d_B$.
:::

#### Basis for Composite Space

If $\{|i\rangle_A : i=1,\ldots,d_A\}$ is a basis for $\mathcal{H}_A$ and $\{|j\rangle_B : j=1,\ldots,d_B\}$ is a basis for $\mathcal{H}_B$, then a basis for $\mathcal{H}_{AB}$ is:

$$\{|i\rangle_A \otimes |j\rangle_B\}_{i,j}

$$ (eq-composite-basis)

We often abbreviate $|i\rangle_A \otimes |j\rangle_B$ as $|ij\rangle$ or $|i\rangle|j\rangle$.

#### Tensor Product of Vectors

For state vectors $|\psi\rangle_A \in \mathcal{H}_A$ and $|\phi\rangle_B \in \mathcal{H}_B$:

$$|\psi\rangle_A \otimes |\phi\rangle_B = \begin{pmatrix} \psi_1 \\ \vdots \\ \psi_{d_A} \end{pmatrix} \otimes \begin{pmatrix} \phi_1 \\ \vdots \\ \phi_{d_B} \end{pmatrix} = \begin{pmatrix} \psi_1 \phi_1 \\ \psi_1 \phi_2 \\ \vdots \\ \psi_{d_A} \phi_{d_B} \end{pmatrix}

$$ (eq-composite-tensor-vector)

Dimension: $d_A \cdot d_B$ components.

#### Tensor Product of Operators

For operators $A \in \text{End}(\mathcal{H}_A)$ and $B \in \text{End}(\mathcal{H}_B)$, the tensor product is:

$$A \otimes B = \begin{pmatrix} a_{11} B & a_{12} B & \cdots \\ a_{21} B & a_{22} B & \cdots \\ \vdots & \vdots & \ddots \end{pmatrix}

$$ (eq-composite-operator-tensor)

where $a_{ij}$ are matrix elements of $A$. Results:- $(A \otimes B)(A' \otimes B') = (AA') \otimes (BB')$ (composition)- $(A \otimes B)^\dagger = A^\dagger \otimes B^\dagger$ (adjoint)- $\text{Tr}(A \otimes B) = \text{Tr}(A) \cdot \text{Tr}(B)$ (trace factorizes)

### Bipartite Systems and Product States

For **two qubits** (A and B), a general state is:

$$|\Psi\rangle_{AB} = \sum_{i,j=0}^{1} c_{ij} |i\rangle_A |j\rangle_B = c_{00}|{\uparrow\uparrow}\rangle + c_{01}|{\uparrow\downarrow}\rangle + c_{10}|{\downarrow\uparrow}\rangle + c_{11}|{\downarrow\downarrow}\rangle

$$ (eq-composite-general-two-qubit)

with normalization $\sum_{i,j}|c_{ij}|^2 = 1$ (4 parameters, minus 1 phase = 7 real degrees of freedom).

#### Product States

:::{admonition} Definition: Separable (Product) States
:class: important

A state is **separable** (or **factorizable**) if it can be written as:

$$|\Psi\rangle_{AB} = |\psi\rangle_A \otimes |\phi\rangle_B

$$ (eq-composite-product-state)
:::

Expanding with $|\psi\rangle_A = \alpha |{\uparrow}\rangle + \beta |{\downarrow}\rangle$ and $|\phi\rangle_B = \gamma |{\uparrow}\rangle + \delta |{\downarrow}\rangle$:

$$c_{00} = \alpha\gamma, \quad c_{01} = \alpha\delta, \quad c_{10} = \beta\gamma, \quad c_{11} = \beta\delta

$$ (eq-composite-product-coefficients)

**Key constraint**: $c_{00} c_{11} = c_{01} c_{10}$ (the determinant of the $2 \times 2$ matrix $[c_{ij}]$ vanishes).

:::**Example**: $|{\uparrow\downarrow}\rangle$ is a product state ($\alpha=1, \beta=0, \gamma=0, \delta=1$); clearly $c_{00}c_{11} = 0 \cdot 0 = 0 = c_{01}c_{10}$.

### Recognizing Entanglement by the Determinant Test

For a two-qubit state $|\Psi\rangle = \sum_{i,j}c_{ij}|ij\rangle$, form the matrix:

$$C = \begin{pmatrix} c_{00} & c_{01} \\ c_{10} & c_{11} \end{pmatrix}

$$ (eq-composite-coefficient-matrix)

- **$\det(C) = 0$**: State is **separable** (product state).
- **$\det(C) \neq 0$**: State is **entangled** (cannot factor).

  **Proof**: For $|\Psi\rangle = (\alpha_0|0\rangle + \alpha_1|1\rangle)_A \otimes (\beta_0|0\rangle + \beta_1|1\rangle)_B$, the coefficient matrix is $C = \boldsymbol{\alpha} \boldsymbol{\beta}^T$, which has rank 1, hence $\det(C) = 0$. Conversely, a rank-1 matrix factors as an outer product.

### Partial Trace and Reduced Density Matrix

Given a bipartite state $\rho_{AB}$, the reduced density matrix for subsystem A is $\rho_A = \text{Tr}_B(\rho_{AB})$, obtained by the **partial trace** operation. This extracts all information about subsystem A alone and is essential for understanding entanglement. See §6.1.2 for the full mathematical treatment.

### Schmidt Decomposition

:::{admonition} Theorem: Schmidt Decomposition
:class: important

Any bipartite pure state $|\Psi\rangle_{AB}$ can be written in the form:

$$|\Psi\rangle_{AB} = \sum_{k=1}^r \lambda_k |u_k\rangle_A \otimes |v_k\rangle_B

$$ (eq-composite-schmidt-decomp)

where:
- $r = \text{rank}(C)$ is the **Schmidt rank** (number of nonzero terms).
- $\lambda_k > 0$ are the **Schmidt coefficients** (normalized: $\sum_k \lambda_k^2 = 1$).
- $\{|u_k\rangle_A\}$ and $\{|v_k\rangle_B\}$ are **orthonormal** sets in their respective spaces.
:::

To decompose a bipartite state, we use the singular value decomposition (SVD):

:::{admonition} Proof: Schmidt Decomposition
:class: dropdown note

Expand $|\Psi\rangle = \sum_{i,j} c_{ij} |i\rangle_A |j\rangle_B$. Form the matrix $C$ with elements $c_{ij}$. By **singular value decomposition**:

$$C = U \Lambda V^\dagger

$$ (eq-composite-svd)

where $U$ ($d_A \times d_A$) and $V$ ($d_B \times d_B$) are unitary, and $\Lambda$ is diagonal with non-negative entries $\lambda_1 \geq \lambda_2 \geq \cdots \geq 0$. Then:

$$|\Psi\rangle = \sum_{k} \lambda_k |u_k\rangle_A |v_k\rangle_B

$$ (eq-composite-svd-result)

where $|u_k\rangle_A$ and $|v_k\rangle_B$ are columns of $U$ and $V$.
:::

:::{admonition} Example: Bell State
:class: example

Consider $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|{\uparrow\uparrow}\rangle + |{\downarrow\downarrow}\rangle)$. Coefficient matrix: $C = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$. SVD: $\Lambda = \frac{1}{\sqrt{2}}I$, so $\lambda_1 = \lambda_2 = \frac{1}{\sqrt{2}}$. Thus $|\Phi^+\rangle$ has **Schmidt rank 2** (maximally entangled for two qubits). Schmidt decomposition: $|\Phi^+\rangle = \frac{1}{\sqrt{2}}|{\uparrow}\rangle_A|{\uparrow}\rangle_B + \frac{1}{\sqrt{2}}|{\downarrow}\rangle_A|{\downarrow}\rangle_B$.
:::
### Reduced Density Matrix and Schmidt Coefficients

From the Schmidt decomposition, the reduced density matrix of subsystem A is:

$$\rho_A = \text{Tr}_B(|\Psi\rangle\langle\Psi|) = \sum_k \lambda_k^2 |u_k\rangle_A\langle u_k|

$$ (eq-composite-reduced-schmidt)

**Key insight**: The eigenvalues of $\rho_A$ are $\lambda_k^2$ (squares of Schmidt coefficients), and the eigenstates are the Schmidt basis states $|u_k\rangle_A$.

#### Purity and Schmidt Rank

$$\text{Tr}(\rho_A^2) = \sum_k \lambda_k^4

$$ (eq-composite-purity-schmidt)

- **Pure state** ($|\Psi\rangle_{AB}$): $\text{Tr}(\rho_A^2) = 1$ iff $\rho_A$ is pure, which occurs iff Schmidt rank = 1 (product state).
- **Maximally entangled** (equal Schmidt coefficients): $\text{Tr}(\rho_A^2) = 1/r$ (minimum purity).

### Schmidt Number as Entanglement Measure

:::{admonition} Schmidt Rank and Entanglement
:class: important

The **Schmidt rank** (number of nonzero Schmidt coefficients) quantifies entanglement:

$$r = \text{rank}(\rho_A) = \dim(\text{supp}(\rho_A))

$$ (eq-composite-schmidt-rank)

- **$r = 1$**: Separable (product state); $\lambda_1 = 1$, all others zero.
- **$r = \min(d_A, d_B)$**: Maximally entangled; all nonzero Schmidt coefficients equal: $\lambda_k = 1/\sqrt{r}$ for $k=1,\ldots,r$.
- **$1 < r < \min(d_A, d_B)$**: Partially entangled.
:::

#### Schmidt Number as a Discrete Entanglement Witness

If $r \geq 2$, the state is **entangled** (cannot be written as a product).**Computational check**: Compute the rank of the coefficient matrix $C = [c_{ij}]$. If $\text{rank}(C) \geq 2$, state is entangled.

### Multipartite Systems

For $N$ systems, the Hilbert space is:

$$\mathcal{H}_{1\cdots N} = \mathcal{H}_1 \otimes \cdots \otimes \mathcal{H}_N

$$ (eq-composite-multipartite)

with dimension $\prod_i d_i$.**Entanglement types**:
- **Fully separable**: $|\Psi\rangle = |\psi_1\rangle \otimes \cdots \otimes |\psi_N\rangle$.
- **Bipartite entanglement**: Entangled between groups of systems (e.g., $|\psi_{1,2}\rangle \otimes |\psi_3\rangle$).
- **Multipartite entanglement**: All systems entangled together; cannot be split into separable blocks.


:::{admonition} Example: GHZ state (3 qubits):
:class: example

$$|\text{GHZ}\rangle = \frac{1}{\sqrt{2}}(|{\uparrow\uparrow\uparrow}\rangle + |{\downarrow\downarrow\downarrow}\rangle)

$$

This is **tripartitely entangled**; removing any one qubit leaves the other two in a product state.
:::


:::{admonition} Entanglement is the rule, not the exception
:class: note

Most states in a composite Hilbert space $\mathcal{H}_A \otimes \mathcal{H}_B$ are entangled. Product (separable) states form a measure-zero subset. Any interaction between subsystems generically creates entanglement.
:::
### Summary
- **Tensor product** $\mathcal{H}_A \otimes \mathcal{H}_B$: Hilbert space of composite system; dimension $d_A \cdot d_B$.
- **Product state**: Separable as $|\psi\rangle_A \otimes |\phi\rangle_B$; characterized by $\det(C) = 0$ (determinant of coefficient matrix).
- **Partial trace**: $\rho_A = \text{Tr}_B(\rho_{AB})$ extracts subsystem A information; eigenvalues = $\lambda_k^2$ (Schmidt coefficients squared); see §6.1.2.
- **Schmidt decomposition**: $|\Psi\rangle = \sum_k \lambda_k |u_k\rangle_A |v_k\rangle_B$; obtained via SVD of coefficient matrix.
- **Schmidt rank** $r = \text{rank}(C)$:   - $r=1$ ⟹ separable (product state).  - $r \geq 2$ ⟹ entangled.  - $r = \min(d_A, d_B)$ ⟹ maximally entangled.
- **Purity**: $\text{Tr}(\rho_A^2) = \sum_k \lambda_k^4 \leq 1$; equals 1 iff $r=1$ (product state).
- **Multipartite entanglement**: More complex; requires analysis of entanglement structure for each bipartition.
