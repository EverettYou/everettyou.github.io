## Lecture Notes

Electron spin is a purely quantum mechanical property with no classical analog, yet it emerges naturally from relativity and the Dirac equation. This section explores how spin manifests as different representations of angular momentum and why spin-1/2 particles like electrons are ubiquitous in nature.

:::{admonition} Spin: Definition and Properties
:class: important

**Spin** is an intrinsic form of angular momentum carried by quantum particles. It is **not** due to the particle spinning (which would violate relativistic constraints) but is a purely quantum mechanical property arising from representation theory of the rotation group.

**Key properties:**
- Spin has **no classical analog** — there is no way to understand it as literal rotation of a body.
- It appears in the relativistic Dirac equation naturally when describing electrons.
- The **magnitude** and **measurable components** are quantized, just like orbital angular momentum.
- The spin quantum number $s$ can be integer or half-integer, unlike orbital angular momentum $\ell$ (which is always integer).
- Spin exists even for point particles with no spatial extent.
:::

### Spin and the Spin-Statistics Theorem

The **spin-statistics theorem** (a cornerstone of quantum field theory) states:

$$


\boxed{\text{Integer spin} \Rightarrow \text{Boson (symmetric wavefunction)}}

$$


$$

\boxed{\text{Half-integer spin} \Rightarrow \text{Fermion (antisymmetric wavefunction)}}

$$


:::{admonition} The Spin-Statistics Connection
:class: important

The link between intrinsic angular momentum and particle statistics is deep and fundamental. It follows from the requirements of **Lorentz invariance** and **causality** in relativistic quantum field theory. This theorem has profound consequences: it explains why electrons are fermions, why photons are bosons, and why the universe has the structure it does.
:::

:::{admonition} Discussion
:class: dropdown warning

**The 4π rotation paradox:** A spin-1/2 particle in state $|\uparrow\rangle$ is rotated by $2\pi$ (one full rotation): $|\uparrow\rangle \to R_z(2\pi)|\uparrow\rangle = -|\uparrow\rangle$. A second $2\pi$ rotation gives: $R_z(2\pi)(-|\uparrow\rangle) = -R_z(2\pi)|\uparrow\rangle = -(-|\uparrow\rangle) = |\uparrow\rangle$. So a **4π total rotation** is needed to return the state. Is this minus sign physically observable, or only meaningful through interference? Can we measure whether a spin-1/2 particle has rotated by $2\pi$ or $0$, or only whether it has rotated by $2\pi$ versus $4\pi$? If global phase is unobservable, why does the 4π structure matter for the validity of quantum mechanics?
:::

:::{admonition} Discussion: Spinors vs. Vectors
:class: dropdown warning

**Half-Integer Spin and Fermi-Dirac Statistics:** The spin-statistics theorem claims that half-integer-spin particles must be fermions (obey Pauli exclusion). Why does the 4π rotation structure of spin-1/2 connect to whether two electrons can occupy the same state?

Can you construct a counterexample: a half-integer-spin particle that is NOT a fermion? What goes wrong? Conversely, could there be integer-spin fermions in principle, even though we don't observe them?
:::

### Spin-1/2: The Simplest Case

### Basis States and Spinors

For a particle with spin-1/2 (like an electron or quark), the spin quantum number is $s = 1/2$. The projection onto the $z$-axis can be $m_s = \pm 1/2$.

:::{admonition} Spin-1/2 Basis States
:class: important

**Basis states**:

$$


|\uparrow\rangle = |s = 1/2, m_s = +1/2\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}

$$ (eq-spin-up)

$$


|\downarrow\rangle = |s = 1/2, m_s = -1/2\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}

$$ (eq-spin-down)

These are **spinors**: two-component complex column vectors. Any spin-1/2 state is a superposition:

$$


|\chi\rangle = a |\uparrow\rangle + b |\downarrow\rangle = \begin{pmatrix} a \\ b \end{pmatrix}

$$


with normalization $|a|^2 + |b|^2 = 1$.
:::

### The Pauli Matrices

Spin-1/2 is represented via the **Pauli matrices**:

$$


\hat{\sigma}^x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}, \quad
\hat{\sigma}^y = \begin{pmatrix} 0 & -\mathrm{i} \\ \mathrm{i} & 0 \end{pmatrix}, \quad
\hat{\sigma}^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}

$$ (eq-pauli-matrices-spin-representations)

The spin operators are:

$$


S_x = \frac{\hbar}{2} \hat{\sigma}^x, \quad S_y = \frac{\hbar}{2} \hat{\sigma}^y, \quad S_z = \frac{\hbar}{2} \hat{\sigma}^z

$$ (eq-spin-operators)

:::{admonition} Properties of Pauli Matrices
:class: important

The Pauli matrices satisfy:
- $\sigma_i^2 = \mathbb{I}$ (each is its own inverse)
- $[\hat{\sigma}^i, \hat{\sigma}^j] = 2\mathrm{i} \epsilon_{ijk} \hat{\sigma}^k$ (commutation relations)
- $\{\hat{\sigma}^i, \hat{\sigma}^j\} = 2 \delta_{ij} \mathbb{I}$ (anticommutation relations)
- $\mathrm{tr}(\hat{\sigma}^i) = 0$ (traceless)
- $\det(\hat{\sigma}^i) = -1$ (determinant is $-1$)

These are the generators of SU(2) rotations and underlie all spin-1/2 physics.
:::

:::{admonition} Derivation: Pauli Matrices and Spin Operators
:class: dropdown

The three **Pauli matrices** satisfy the commutation relations:

$$


[\sigma_i, \sigma_j] = 2\mathrm{i}\epsilon_{ijk}\sigma_k

$$


and the anticommutation relations:

$$


\{\sigma_i, \sigma_j\} = 2\delta_{ij}\mathbb{I}

$$


The spin operators are related to the Pauli matrices by $\hat{S}_i = \frac{\hbar}{2}\sigma_i$. The Pauli matrices generate rotations in SU(2), the universal cover of the rotation group SO(3), which explains why spin-1/2 transforms as a spinor.
:::

### Eigenvalues and Eigenvectors

For $S_z = \frac{\hbar}{2} \hat{\sigma}^z$:

$$


S_z |\uparrow\rangle = \frac{\hbar}{2} |\uparrow\rangle, \quad S_z |\downarrow\rangle = -\frac{\hbar}{2} |\downarrow\rangle

$$


The total spin squared:

$$


S^2 = S_x^2 + S_y^2 + S_z^2 = \frac{3\hbar^2}{4} I

$$


Both $|\uparrow\rangle$ and $|\downarrow\rangle$ are eigenstates with eigenvalue $\frac{3\hbar^2}{4} = \hbar^2 s(s+1)$ where $s = 1/2$. ✓

:::{admonition} Derivation: Spin-1/2 Matrix Representation
:class: dropdown

We derive the Pauli matrices from the requirement that spin operators satisfy $[S_i, S_j] = \mathrm{i}\hbar \epsilon_{ijk} S_k$ with $S^i = \frac{\hbar}{2}\hat{\sigma}^i$.

**Step 1: Dimensionality**

For spin-1/2, we have $2j + 1 = 2$ basis states: $|\uparrow\rangle$ and $|\downarrow\rangle$. Operators must be $2 \times 2$ matrices.

**Step 2: $S_z$ eigenvalues**

$S_z$ must have eigenvalues $\pm\frac{\hbar}{2}$ for the two basis states. Diagonal form:

$$


S_z = \frac{\hbar}{2}\begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} = \frac{\hbar}{2}\hat{\sigma}^z

$$


**Step 3: Ladder operators**

From $J_+ |j, m\rangle = \hbar\sqrt{j(j+1) - m(m+1)} |j, m+1\rangle$, with $j = 1/2$:

$$


S_+ |\downarrow\rangle = \hbar\sqrt{(1/2)(3/2) - (-1/2)(1/2)} |\uparrow\rangle = \hbar |\uparrow\rangle

$$


$$

S_+ |\uparrow\rangle = 0

$$


$$

S_- |\uparrow\rangle = \hbar\sqrt{(1/2)(3/2) - (1/2)(1/2)} |\downarrow\rangle = \hbar |\downarrow\rangle

$$


$$

S_- |\downarrow\rangle = 0

$$


Therefore:

$$


S_+ = \hbar\begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}, \quad S_- = \hbar\begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix}

$$


**Step 4: Cartesian components**

$$


S_x = \frac{1}{2}(S_+ + S_-) = \frac{\hbar}{2}\begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} = \frac{\hbar}{2}\hat{\sigma}^x

$$


$$

S_y = \frac{1}{2i}(S_+ - S_-) = \frac{\hbar}{2}\begin{pmatrix} 0 & -\mathrm{i} \\ \mathrm{i} & 0 \end{pmatrix} = \frac{\hbar}{2}\hat{\sigma}^y

$$


**Step 5: Verification**

We can verify $[S_i, S_j] = \mathrm{i}\hbar \epsilon_{ijk} S_k$ directly using the Pauli matrix commutation relations $[\hat{\sigma}^i, \hat{\sigma}^j] = 2\mathrm{i}\epsilon_{ijk}\hat{\sigma}^k$. ✓
:::

### Spinor Rotations and the Double Cover

### How Spinors Transform

Under a rotation by angle $\theta$ about an axis $\hat{n} = (n_x, n_y, n_z)$ with $|\hat{n}| = 1$, a spinor transforms as:

:::{admonition} Spinor Rotation Formula
:class: important

$$


\chi'(\boldsymbol{r}, \theta) = \mathrm{e}^{-\mathrm{i}\theta \hat{\boldsymbol{\sigma}} \cdot \hat{n} / 2} \chi(\boldsymbol{r}, 0) = \mathrm{e}^{-\mathrm{i} S \theta / \hbar} \chi(\boldsymbol{r}, 0)

$$ (eq-spinor-rotation)

**Note the factor of 1/2 in the exponent.** This is the defining feature of spin-1/2.
:::

### The $2\pi$ and $4\pi$ Mystery

A remarkable consequence:

:::{admonition} Derivation: Double-Angle Rotation of Spinors
:class: dropdown

**A $2\pi$ rotation**:

$$


\mathrm{e}^{-\mathrm{i} \pi \hat{\sigma}^z} = \mathrm{e}^{-\mathrm{i} \pi \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}} = \begin{pmatrix} -1 & 0 \\ 0 & -1 \end{pmatrix} = -I

$$


The spinor picks up a phase of **$-1$** (or equivalently, $\mathrm{e}^{\mathrm{i}\pi}$). The spinor returns to its original state **after only a $2\pi$ rotation**, but the overall phase changes.

**A $4\pi$ rotation**:

$$


\mathrm{e}^{-\mathrm{i} 2\pi \hat{\sigma}^z} = (\mathrm{e}^{-\mathrm{i} \pi \hat{\sigma}^z})^2 = (-I)^2 = I

$$


Only after a **$4\pi$ rotation** does the spinor return to its original form **with phase $+1$**.
:::

:::{admonition} Physical Meaning of the $2\pi$ Minus Sign
:class: important

This is **not** unphysical! Remember:
- **Quantum states are projective**: phase factors of $\pm 1$ are global and unobservable in simple measurements.
- However, **relative phases between components matter** and can be measured through interference.
- After a $2\pi$ rotation, all physical observables (transition probabilities, expectation values) return to their original values because $\chi'$ and $-\chi'$ represent the same quantum state.

Yet there is a deep topological significance: Spinors form a **double cover** of rotations—the group $SU(2)$ (acting on spinors) is a double cover of the rotation group $SO(3)$ (acting on vectors).
:::

### Spin-1: Three-Dimensional Representation

### Basis States

For spin-1 particles ($s = 1$), $m_s \in \{-1, 0, +1\}$ and the Hilbert space is **three-dimensional**:

$$


|1, +1\rangle = \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}, \quad
|1, 0\rangle = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}, \quad
|1, -1\rangle = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}

$$

:::{admonition} Spin-1 Operators
:class: important

$$


S_z = \hbar \begin{pmatrix} 1 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & -1 \end{pmatrix}

$$


$$

S_+ = \hbar \begin{pmatrix} 0 & \sqrt{2} & 0 \\ 0 & 0 & \sqrt{2} \\ 0 & 0 & 0 \end{pmatrix}, \quad
S_- = \hbar \begin{pmatrix} 0 & 0 & 0 \\ \sqrt{2} & 0 & 0 \\ 0 & \sqrt{2} & 0 \end{pmatrix}

$$


$$

S_x = \frac{\hbar}{\sqrt{2}} \begin{pmatrix} 0 & 1 & 0 \\ 1 & 0 & 1 \\ 0 & 1 & 0 \end{pmatrix}, \quad
S_y = \frac{\hbar}{\sqrt{2}} \begin{pmatrix} 0 & -i & 0 \\ i & 0 & -\mathrm{i} \\ 0 & \mathrm{i} & 0 \end{pmatrix}

$$
:::

:::{admonition} Example: Particles with Spin-1
:class: example

- **Photons**: spin-1 bosons (massless, move at light speed).
- **W and Z bosons**: spin-1, massive, mediate weak nuclear force.
- **Pions**: spin-0 bosons (not spin-1, just for comparison).
:::

:::{admonition} Derivation: Spin-1 Matrix Representation
:class: dropdown

For spin-1, we have $2j + 1 = 3$ basis states: $|1, +1\rangle$, $|1, 0\rangle$, $|1, -1\rangle$.

**Step 1: $S_z$ eigenvalues**

The basis is ordered as $|+1\rangle, |0\rangle, |-1\rangle$ with eigenvalues $\hbar, 0, -\hbar$:

$$


S_z = \hbar\begin{pmatrix} 1 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & -1 \end{pmatrix}

$$


**Step 2: Ladder operators**

From $J_+ |j, m\rangle = \hbar\sqrt{j(j+1) - m(m+1)} |j, m+1\rangle$ with $j = 1$:

$$


S_+ |1, -1\rangle = \hbar\sqrt{2 - (-1)(0)} |1, 0\rangle = \hbar\sqrt{2} |1, 0\rangle

$$


$$

S_+ |1, 0\rangle = \hbar\sqrt{2 - 0(1)} |1, +1\rangle = \hbar\sqrt{2} |1, +1\rangle

$$


$$

S_+ |1, +1\rangle = 0

$$


In matrix form:

$$


S_+ = \hbar\begin{pmatrix} 0 & \sqrt{2} & 0 \\ 0 & 0 & \sqrt{2} \\ 0 & 0 & 0 \end{pmatrix}

$$


Similarly for $S_-$:

$$


S_- = \hbar\begin{pmatrix} 0 & 0 & 0 \\ \sqrt{2} & 0 & 0 \\ 0 & \sqrt{2} & 0 \end{pmatrix}

$$


**Step 3: Cartesian components**

$$


S_x = \frac{1}{2}(S_+ + S_-) = \frac{\hbar}{\sqrt{2}}\begin{pmatrix} 0 & 1 & 0 \\ 1 & 0 & 1 \\ 0 & 1 & 0 \end{pmatrix}

$$


$$

S_y = \frac{1}{2i}(S_+ - S_-) = \frac{\hbar}{\sqrt{2}}\begin{pmatrix} 0 & -i & 0 \\ i & 0 & -\mathrm{i} \\ 0 & \mathrm{i} & 0 \end{pmatrix}

$$


These satisfy the spin-1 angular momentum algebra. ✓
:::

### Rotations of Spin-1

A spin-1 object rotates like a **classical vector**. A $2\pi$ rotation returns it to its original state (no $-1$ phase).

### General Spin-$j$ Representation

### Dimensions and States

For arbitrary angular momentum quantum number $j \in \{0, 1/2, 1, 3/2, \ldots\}$:

- **Dimension**: $(2j+1)$-dimensional Hilbert space.
- **Basis states**: $|j, m\rangle$ for $m \in \{-j, -j+1, \ldots, j-1, j\}$ (total of $2j+1$ states).

### Matrix Representations

The spin operators are $(2j+1) \times (2j+1)$ matrices. Matrix elements are:

$$


\langle j, m | S_z | j, m' \rangle = \hbar m \delta_{mm'}

$$


$$

\langle j, m | S_+ | j, m' \rangle = \hbar \sqrt{j(j+1) - m'(m'+1)} \delta_{m, m'+1}

$$


These can be explicitly computed and tabulated (or generated numerically).

### Rotation Operator

The general rotation operator for spin-$j$ is:

$$


R(\theta, \hat{n}) = \mathrm{e}^{-\mathrm{i} \theta \boldsymbol{S} \cdot \hat{n} / \hbar}

$$


For **integer** $j$: a $2\pi$ rotation gives $\mathrm{e}^{-\mathrm{i} 2\pi j} = 1$ (returns unchanged).

For **half-integer** $j$: a $2\pi$ rotation gives $\mathrm{e}^{-\mathrm{i} 2\pi j} = -1$ (picks up $-1$ phase).

This is how the spin-statistics theorem emerges: half-integer spins naturally pick up minus signs under rotations, leading to antisymmetric (fermionic) wavefunctions.

### Combining Spatial and Spin Degrees of Freedom

### Single-Particle Wavefunction

A single particle has both **spatial** ($\boldsymbol{r}$) and **spin** ($s_z$) degrees of freedom. The full wavefunction is:

:::{admonition} Spinor Wavefunction
:class: important

$$


\psi(\boldsymbol{r}, s_z) = \psi_{\uparrow}(\boldsymbol{r}) |\uparrow\rangle + \psi_{\downarrow}(\boldsymbol{r}) |\downarrow\rangle

$$ (eq-spinor-wavefunction)

This is a **two-component wavefunction** or **spinor wavefunction**. At each point $\boldsymbol{r}$ in space, the value is a two-component spinor.
:::

### Probability Density

The total probability of finding the particle in state $\boldsymbol{r}$ is:

$$


\rho(\boldsymbol{r}) = |\psi_{\uparrow}(\boldsymbol{r})|^2 + |\psi_{\downarrow}(\boldsymbol{r})|^2

$$


The **spin density** (probability of finding spin-up or spin-down at $\boldsymbol{r}$):

$$


\rho_{\uparrow}(\boldsymbol{r}) = |\psi_{\uparrow}(\boldsymbol{r})|^2, \quad \rho_{\downarrow}(\boldsymbol{r}) = |\psi_{\downarrow}(\boldsymbol{r})|^2

$$

### Expectation Values

The expectation value of spin $S_z$:

$$


\langle S_z \rangle = \frac{\hbar}{2} \int d^3r \left( |\psi_{\uparrow}(\boldsymbol{r})|^2 - |\psi_{\downarrow}(\boldsymbol{r})|^2 \right)

$$

### Product States vs. Entangled States

A **product state** factors into spatial and spin parts:

$$


\psi(\boldsymbol{r}, s_z) = \phi(\boldsymbol{r}) \chi(s_z)

$$


For example, an electron in the ground state of hydrogen with spin-up:

$$


\psi_{1s, \uparrow}(\boldsymbol{r}, s_z) = \phi_{1s}(\boldsymbol{r}) |\uparrow\rangle

$$


**Entangled states** cannot be factored. Example:

$$


\psi(\boldsymbol{r}, s_z) = \frac{1}{\sqrt{2}} \phi_1(\boldsymbol{r}) |\uparrow\rangle + \frac{1}{\sqrt{2}} \phi_2(\boldsymbol{r}) |\downarrow\rangle

$$


Here the spatial part is entangled with the spin part: the particle is in different locations depending on its spin state.

### Spin Measurement and the Stern-Gerlach Experiment

### Classical Expectation

If spin were a classical angular momentum vector, a Stern-Gerlach apparatus (which measures spin along $z$) applied to a spin-1/2 particle would give a **continuous range** of outcomes, depending on the angle of the spin vector.

### Quantum Reality

In reality, we observe only **two outcomes**:

- **Spin-up** ($m_s = +1/2$): measured with probability $|a|^2$
- **Spin-down** ($m_s = -1/2$): measured with probability $|b|^2$

where $|\chi\rangle = a|\uparrow\rangle + b|\downarrow\rangle$.

:::{admonition} Discussion: Sequential Spin Measurements
:class: dropdown warning

A remarkable feature of quantum spin measurement:

1. Prepare spin-up along $z$: $|\uparrow\rangle_z$
2. Measure along $x$: get $|\uparrow\rangle_x$ or $|\downarrow\rangle_x$ with equal probability.
3. If we got $|\uparrow\rangle_x$, measure along $z$ again: **no longer guaranteed to be spin-up**!

This shows that spin measurement **disturbs the state** and that the spin along different axes are **incompatible observables** (they don't commute). What does this tell us about the nature of measurement in quantum mechanics? Does the act of measuring necessarily collapse the wavefunction, or could there be hidden information about the spin state before measurement?
:::

### Summary

- **Spin** is intrinsic angular momentum with no classical analog; it appears naturally in relativistic quantum mechanics.
- **Spin-1/2**: basis states $|\uparrow\rangle, |\downarrow\rangle$; operators are $\frac{\hbar}{2}\hat{\boldsymbol{\sigma}}$ (Pauli matrices).
- **Spinors** are two-component column vectors; they transform with **half angles** under rotations.
- **$2\pi$ vs. $4\pi$ rotations**: Spin-1/2 picks up $-1$ phase after $2\pi$ rotation (returns to original with $4\pi$); integer spins don't.
- **General spin-$j$**: $(2j+1)$-dimensional representation; integer $j$ are bosons, half-integer $j$ are fermions.
- **Spatial-spin coupling**: Full wavefunction $\psi(\boldsymbol{r}, s_z)$ is a spinor at each point in space.
- **Measurement**: Only discrete outcomes; sequential measurements along different axes disturb the state.
