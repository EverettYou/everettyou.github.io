## Lecture Notes

When quantum systems combine, their state space is not a simple union—it is a tensor product that encodes all possible correlations between subsystems. This section introduces the Hilbert space of composite quantum systems and reveals how entanglement arises as a fundamental feature of multi-qubit systems.
### Combining Two Single-Qubit Systems

Each qubit has a 2-dimensional Hilbert space with basis states $|0\rangle$ and $|1\rangle$.

The two-qubit Hilbert space has dimension $2 \times 2 = 4$. The basis states are formed by tensoring single-qubit states:

$$

|00\rangle = |0\rangle_A \otimes |0\rangle_B, \quad
|01\rangle = |0\rangle_A \otimes |1\rangle_B, \quad
|10\rangle = |1\rangle_A \otimes |0\rangle_B, \quad
|11\rangle = |1\rangle_A \otimes |1\rangle_B

$$

More compactly, we label two-qubit states as $|ij\rangle$ where $i,j \in \{0,1\}$, or sometimes just write $|i\rangle \otimes |j\rangle$ to emphasize the tensor product structure.

**Orthonormality**:

$$

\langle i_1 j_1 | i_2 j_2 \rangle = \langle i_1 | i_2 \rangle \langle j_1 | j_2 \rangle = \delta_{i_1 i_2} \delta_{j_1 j_2}

$$ (eq-two-qubit-orthonormal)

### General Case: $N$ Particles with $d$ States Each

For $N$ quantum particles, each having access to $d$ single-particle basis states $|\alpha\rangle$ (where $\alpha = 1, 2, \ldots, d$), the combined Hilbert space is:

$$

\mathcal{H}_1 \otimes \mathcal{H}_2 \otimes \cdots \otimes \mathcal{H}_N

$$

This space has dimension:

$$

\dim(\mathcal{H}_{\text{total}}) = d^N

$$ (eq-hilbert-dimension)

A complete orthonormal basis consists of all product states:

$$

|\alpha_1\rangle |\alpha_2\rangle \cdots |\alpha_N\rangle = |\alpha_1 \alpha_2 \ldots \alpha_N\rangle

$$

where each $\alpha_i \in \{1, 2, \ldots, d\}$.

:::{admonition} Exponential Growth and Classical Limits
:class: note

The dimension grows **exponentially** with $N$. For $N = 10$ particles with $d = 2$ (qubits), the total Hilbert space has $2^{10} = 1024$ dimensions. For $N = 300$ and $d = 2$, we get $2^{300} \approx 10^{90}$ dimensions—larger than the number of atoms in the universe! This exponential growth is both a source of quantum computing's power and a reason why classical simulation of quantum systems becomes intractable.
:::

:class: note

We use several equivalent notations: |α₁α₂...αₙ⟩, |α₁⟩|α₂⟩···|αₙ⟩, and |α₁⟩⊗|α₂⟩⊗···⊗|αₙ⟩. The abbreviated form is standard when indices don't overlap.
:::

:class: caution

Not all states in ℋ₁⊗ℋ₂ are product states. Bell states like |Φ⁺⟩ = 1/√2(|00⟩+|11⟩) cannot be factored and exhibit maximal correlations upon measurement.
:::


## Product States and Entanglement

### Product States

A **product state** (or separable state) is a state that can be written as a tensor product of single-particle states:

$$

|\psi\rangle = |\psi_1\rangle \otimes |\psi_2\rangle \otimes \cdots \otimes |\psi_N\rangle

$$ (eq-product-state)

For a product state:
- **Particle independence**: measurement on particle $i$ does not affect the state of particle $j$ ($j \neq i$).
- **Factorizable expectation values**: $\langle A_1 \otimes B_2 \rangle = \langle A_1 \rangle \langle B_2 \rangle$ where $A_1$ acts on particle 1 and $B_2$ acts on particle 2.

### Entangled States

Most states in the tensor product space **cannot** be written as a simple product. These are **entangled states**.

:::{admonition} Example: The Bell State (Two Qubits)
:class: example


$$

|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)

$$

This state **cannot** be written as $|\psi_A\rangle \otimes |\psi_B\rangle$ for any $|\psi_A\rangle, |\psi_B\rangle$. If we measure qubit A and find it in state $|0\rangle$, then qubit B is **definitely** in state $|0\rangle$ (not a superposition). The measurement on A instantaneously affects the state of B—perfect correlation.
:::
### General Multi-Particle State

Any state in the $N$-particle Hilbert space can be expanded as:

$$

|\psi\rangle = \sum_{\alpha_1=1}^d \sum_{\alpha_2=1}^d \cdots \sum_{\alpha_N=1}^d c_{\alpha_1 \alpha_2 \ldots \alpha_N} |\alpha_1 \alpha_2 \ldots \alpha_N\rangle

$$ (eq-general-state-expansion)

The coefficients $c_{\alpha_1 \alpha_2 \ldots \alpha_N}$ are complex amplitudes, with normalization $\sum_{\alpha_1, \ldots, \alpha_N} |c_{\alpha_1 \ldots \alpha_N}|^2 = 1$.

## Operators on Multi-Particle Systems

### Single-Particle Operators

An operator that acts on only one particle is extended to the full space by tensoring with identity on other particles.

For particle $i$ with operator $A_i$:

$$

\hat{A}_i = \mathbb{I}_1 \otimes \cdots \otimes \mathbb{I}_{i-1} \otimes A_i \otimes \mathbb{I}_{i+1} \otimes \cdots \otimes \mathbb{I}_N

$$ (eq-single-particle-op)

**Action on product states**:

$$

\hat{A}_i |\alpha_1 \alpha_2 \ldots \alpha_N\rangle = |\alpha_1 \ldots \alpha_{i-1}\rangle (A_i |\alpha_i\rangle) |\alpha_{i+1} \ldots \alpha_N\rangle

$$

**Example**: For a two-qubit system, the Pauli $X$ operator acting on qubit 2:

$$

X_2 = I \otimes X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \otimes I

$$

$$

X_2 |01\rangle = |0\rangle \otimes (X|1\rangle) = |0\rangle \otimes |0\rangle = |00\rangle

$$

### Two-Body Interactions

An operator acting on pairs of particles takes the form:

$$

V_{ij} = \mathbb{I}_1 \otimes \cdots \otimes V_{ij} \otimes \cdots \otimes \mathbb{I}_N

$$

where $V_{ij}$ couples particles $i$ and $j$.

:::{admonition} Example: Spin-spin coupling
:class: example


$$

H_{int} = J \boldsymbol{S}_1 \cdot \boldsymbol{S}_2

$$

This couples the spins of particles 1 and 2, energetically favoring or disfavoring alignment depending on the sign of $J$.
:::
### Commutation of Single-Particle Operators

Operators acting on different particles **always commute**:

$$

[A_i, B_j] = 0 \quad \text{for } i \neq j

$$ (eq-different-particle-commute)

This is because they act on orthogonal subspaces of the tensor product.

## Basis States and Notation Conventions

### Standard Notation

For an $N$-particle system, we use several interchangeable notations:

$$

|\alpha_1 \alpha_2 \ldots \alpha_N\rangle = |\alpha_1\rangle |\alpha_2\rangle \cdots |\alpha_N\rangle = |\alpha_1\rangle \otimes |\alpha_2\rangle \otimes \cdots \otimes |\alpha_N\rangle

$$

The commas are often omitted for brevity when there is no ambiguity.

### Position and Momentum Basis

For continuous quantum systems (particles in space), basis states are labeled by position or momentum:

$$

|\boldsymbol{r}_1 \boldsymbol{r}_2 \ldots \boldsymbol{r}_N\rangle \quad \text{(position basis)}

$$

A many-body wavefunction is:

$$

\psi(\boldsymbol{r}_1, \boldsymbol{r}_2, \ldots, \boldsymbol{r}_N) = \langle \boldsymbol{r}_1 \boldsymbol{r}_2 \ldots \boldsymbol{r}_N | \psi \rangle

$$

### Discrete Basis (Modes)

For systems with discrete single-particle spectrum (e.g., trapped particles, cavity modes), label basis states by mode indices:

$$

|\phi_i\rangle \quad \text{for } i = 1, 2, 3, \ldots

$$

A two-particle state might be:

$$

|\phi_i \phi_j\rangle = |i, j\rangle

$$

:::{admonition} Example: Two-Particle Hilbert Space
:class: example


### Two Distinguishable Qubits

For two distinguishable particles (labeled A and B), each a qubit, the complete basis is:

$$

\{|00\rangle, |01\rangle, |10\rangle, |11\rangle\}

$$

Any two-qubit state:

$$

|\psi\rangle = c_{00} |00\rangle + c_{01} |01\rangle + c_{10} |10\rangle + c_{11} |11\rangle

$$

with $|c_{00}|^2 + |c_{01}|^2 + |c_{10}|^2 + |c_{11}|^2 = 1$.

### Product State Example

The state $|\psi_A\rangle \otimes |\psi_B\rangle = \left(\frac{|0\rangle + |1\rangle}{\sqrt{2}}\right)_A \otimes |0\rangle_B$ expands as:

$$

|\psi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |10\rangle)

$$

This factorizes: knowing particle A's state constrains particle B to $|0\rangle$ always.

### Entangled State Example

The Bell state $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$ has a beautiful property: measuring either qubit in the computational basis collapses both.

- If we measure qubit A and get outcome "0", qubit B **must** be in state $|0\rangle$.
- If we measure qubit A and get outcome "1", qubit B **must** be in state $|1\rangle$.

The state cannot be factored into independent pieces; the qubits are **maximally correlated**.
:::

## Key Insights

### Hilbert Space Dimension

The exponential growth $d^N$ means that:
- Specifying a general $N$-particle quantum state requires $O(d^N)$ complex numbers.
- This exponential complexity is why classical computers struggle to simulate quantum systems but also why quantum computers can be powerful.

### Distinguishability vs. Indistinguishability

The tensor product as formulated here assumes **distinguishable** particles (e.g., "particle A" and "particle B" are labeled).

For **identical** particles, not all states in $\mathcal{H}_1 \otimes \mathcal{H}_2 \otimes \cdots$ are physical. Only symmetric (bosons) or antisymmetric (fermions) combinations are allowed. This dramatically reduces the effective Hilbert space dimension.

### Entanglement as a Computational Resource

Entanglement—correlations that cannot be explained by local product states—is a key resource in quantum computing and quantum information. It allows:
- Quantum cryptography (secure key distribution).
- Quantum teleportation (moving a quantum state).
- Exponential speedups in some algorithms (quantum computers).

## Summary

- **Tensor product** $\mathcal{H}_A \otimes \mathcal{H}_B$ combines single-particle Hilbert spaces.
- **Dimension scaling**: An $N$-particle system with $d$ single-particle states has dimension $d^N$.
- **Basis states**: Form complete set by tensoring all single-particle basis state combinations.
- **Product states** factor into single-particle states; most states are **entangled**.
- **Single-particle operators** act via identity on other particles; different-particle operators commute.
- **Exponential growth** is foundational to quantum advantage and scaling challenges.
