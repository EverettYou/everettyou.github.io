## Lecture Notes

:::{admonition} Key Concept: Second Quantization
:class: important

Creation and annihilation operators provide a powerful reformulation of quantum mechanics that naturally handles systems with variable particle number and automatically enforces quantum statistics. Instead of asking "which particle is in which state?", we ask "how many particles occupy each state?"
:::

### First-Quantized Picture: "Which Particle in Which State?"

In the **first-quantized** approach, we specify the state of each particle individually:

$$


\text{Particle 1 in state } |\phi_1\rangle, \quad \text{Particle 2 in state } |\phi_2\rangle, \quad \ldots

$$


For a system of $N$ distinguishable particles, the many-body state lives in a tensor product space:

$$


|\psi\rangle = |\phi_1\rangle \otimes |\phi_2\rangle \otimes \cdots \otimes |\phi_N\rangle

$$


For **identical** particles, we must symmetrize or antisymmetrize, making the first-quantized picture awkward and unwieldy.

### Second-Quantized Picture: "How Many Particles in Each State?"

The **second-quantized** approach asks a fundamentally different question:

$$


n_1 \text{ particles in state } 1, \quad n_2 \text{ particles in state } 2, \quad \ldots

$$


A state is specified entirely by **occupation numbers** $\{n_\alpha\}$. This is much more natural for identical particles:

- We never ask "which particle is where?"
- Symmetry/antisymmetry is built in from the start
- Variable particle number is handled elegantly (via the vacuum state)

:::{admonition} Note: Why "Second" Quantization?
:class: note

Despite the name, there is only **one** level of quantization in second quantization. The terminology is historical. What changed is not the quantization level, but the choice of variables: from particle positions to occupation numbers. This reformulation makes particle-number changes natural and encoded in the operator algebra itself.
:::

### Relation to Quantum Field Theory

Second quantization is the language of **quantum field theory**. The fields are:

- $\hat{\psi}^\dagger(\boldsymbol{r})$: **field creation operator** (creates a particle at position $\boldsymbol{r}$)
- $\hat{\psi}(\boldsymbol{r})$: **field annihilation operator** (annihilates a particle at position $\boldsymbol{r}$)

These field operators automatically enforce the correct commutation or anticommutation relations, ensuring the proper statistics (bosonic or fermionic) without explicit symmetrization.

:::{admonition} Discussion: Can We Distinguish Bosonic and Fermionic Vacua?
:class: dropdown warning

The bosonic vacuum $|0\rangle_B$ and fermionic vacuum $|0\rangle_F$ both have zero particles. But they support different excitation spectra: bosons permit arbitrary occupation $n_\alpha = 0, 1, 2, \ldots$, while fermions restrict to $n_\alpha \in \{0, 1\}$.

**Question**: Is the vacuum state genuinely different, or is the difference purely a feature of the operator algebra? Could we tell the two vacua apart by a measurement that *does not involve* creating particles?
:::

### Summary

- **Occupation numbers** $\{n_\alpha\}$ replace particle labels; state $|n_1, n_2, \ldots\rangle$ specifies how many particles occupy each single-particle mode
- **Fock space** $\mathcal{F} = \bigoplus_{N=0}^\infty \mathcal{H}_N$ includes sectors of all particle numbers; vacuum $|0\rangle$ has no particles
- **Creation operator** $\hat{a}_\alpha^\dagger$ adds a particle to mode $\alpha$; **annihilation operator** $\hat{a}_\alpha$ removes one
- **Number operator** $\hat{n}_\alpha = \hat{a}_\alpha^\dagger \hat{a}_\alpha$ counts particles in mode $\alpha$
- Second quantization automatically enforces Pauli exclusion for fermions and boson statistics; simplifies identical-particle calculations dramatically
## Fock Space and Occupation Number Representation

### Definition of a Fock State

A **Fock state** (also called a **number state** or **occupation number state**) is labeled by the occupation numbers of all single-particle states:

$$


|n_1, n_2, \ldots, n_D\rangle \quad \text{or} \quad |\{n_\alpha\}\rangle

$$ (eq-fock-state-label)

where:
- $n_\alpha$ = number of particles in single-particle state $|\alpha\rangle$
- $D$ = total number of single-particle states (dimension of single-particle space)
- $\sum_\alpha n_\alpha = N$ (total particle constraint, though in some formulations $N$ is variable)

### The Vacuum State

The **vacuum state** $|0\rangle$ or $|\text{vac}\rangle$ has no particles:

$$


|0\rangle = |0, 0, \ldots, 0\rangle \quad (n_\alpha = 0 \text{ for all } \alpha)

$$ (eq-vacuum-state)

It is the **ground state** of the Fock space and serves as the reference for building all other states.

### Single-Particle Excitations

States with one particle in state $\alpha$:

$$


|n_\alpha = 1\rangle = |0, 0, \ldots, 0, \underbrace{1}_{\alpha}, 0, \ldots, 0\rangle

$$


These are the **single-particle basis states** expressed in the Fock representation.

### Multi-Particle States

States with multiple particles:

$$


|1, 2, 0, 3, \ldots\rangle \quad \text{means } 1 \text{ particle in state } 1, \, 2 \text{ in state } 2, \, 3 \text{ in state } 4, \, \ldots

$$


### Fock Space Completeness

All Fock states form an **orthonormal basis** for Fock space:

$$


\langle n_1', n_2', \ldots | n_1, n_2, \ldots \rangle = \prod_\alpha \delta_{n_\alpha' n_\alpha}

$$ (eq-fock-orthonormality)

This basis is complete for quantum states involving any number of particles.
## Creation and Annihilation Operators

### Physical Interpretation

- **Creation operator** $b^\dagger_\alpha$ (or $c^\dagger_\alpha$ for fermions): adds a particle to state $\alpha$.
- **Annihilation operator** $b_\alpha$ (or $c_\alpha$ for fermions): removes a particle from state $\alpha$.

### Action on Fock States: Bosons

For bosons, with no restriction on occupation numbers:

$$


b^\dagger_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = \sqrt{n_\alpha + 1} |n_1, \ldots, n_\alpha + 1, \ldots\rangle

$$ (eq-boson-creation)

$$


b_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = \sqrt{n_\alpha} |n_1, \ldots, n_\alpha - 1, \ldots\rangle

$$ (eq-boson-annihilation)

**Key feature**: The $\sqrt{n_\alpha + 1}$ and $\sqrt{n_\alpha}$ factors are **combinatorial weights** arising from the symmetry of bosonic states. They ensure proper normalization.

### Action on Fock States: Fermions

For fermions, occupation is restricted to 0 or 1. Creation has an important modification:

$$


c^\dagger_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = \begin{cases}
(-1)^{P_\alpha} |n_1, \ldots, n_\alpha + 1, \ldots\rangle & \text{if } n_\alpha = 0 \
0 & \text{if } n_\alpha = 1
\end{cases}

$$ (eq-fermion-creation)

where $P_\alpha = \sum_{i < \alpha} n_i$ is the **parity** (number of fermions in states before $\alpha$).

$$


c_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = \begin{cases}
(-1)^{P_\alpha} |n_1, \ldots, n_\alpha - 1, \ldots\rangle & \text{if } n_\alpha = 1 \
0 & \text{if } n_\alpha = 0
\end{cases}

$$ (eq-fermion-annihilation)

**Physical meaning of the sign**: The $(-1)^{P_\alpha}$ phase ensures that the fermionic state remains **antisymmetric** under particle exchange.

:::{admonition} Example: Boson and Fermion Examples
:class: example


**Boson example**:

$$


b^\dagger_1 |0, 3, 0, \ldots\rangle = \sqrt{0 + 1} |1, 3, 0, \ldots\rangle = |1, 3, 0, \ldots\rangle

$$


**Fermion example**:

$$


c^\dagger_1 |0, 1, 1, \ldots\rangle = |1, 1, 1, \ldots\rangle \quad (P_1 = 0 \text{ since state 1 is empty})

$$


$$

c^\dagger_3 |1, 1, 0, \ldots\rangle = (-1)^{1+1} |1, 1, 1, \ldots\rangle = |1, 1, 1, \ldots\rangle \quad (P_3 = 2)

$$


:::
## Commutation and Anticommutation Relations

### Bosonic Relations: Commutators

Creation and annihilation operators for bosons satisfy the **canonical commutation relations**:

$$


[b_\alpha, b^\dagger_\beta] = \delta_{\alpha\beta}, \quad [b_\alpha, b_\beta] = 0, \quad [b^\dagger_\alpha, b^\dagger_\beta] = 0

$$ (eq-boson-commutation)

:::{admonition} Canonical Commutation Relations (Bosons)
:class: important

Creation and annihilation operators for bosons satisfy:

$$


[b_\alpha, b^\dagger_\beta] = \delta_{\alpha\beta}, \quad [b_\alpha, b_\beta] = 0, \quad [b^\dagger_\alpha, b^\dagger_\beta] = 0

$$

where the commutator is $[A, B] = AB - BA$.

**Equivalently**:
- Operators on **different states** commute: $[b_\alpha, b_\beta] = 0$ for $\alpha \neq \beta$.
- The fundamental relation: $[b_\alpha, b^\dagger_\alpha] = 1$ (for the same state).
:::

### Fermionic Relations: Anticommutators

Creation and annihilation operators for fermions satisfy **canonical anticommutation relations**:

$$


\{c_\alpha, c^\dagger_\beta\} = \delta_{\alpha\beta}, \quad \{c_\alpha, c_\beta\} = 0, \quad \{c^\dagger_\alpha, c^\dagger_\beta\} = 0

$$ (eq-fermion-anticommutation)

:::{admonition} Canonical Anticommutation Relations (Fermions)
:class: important

Creation and annihilation operators for fermions satisfy:

$$


\{c_\alpha, c^\dagger_\beta\} = \delta_{\alpha\beta}, \quad \{c_\alpha, c_\beta\} = 0, \quad \{c^\dagger_\alpha, c^\dagger_\beta\} = 0

$$

where the anticommutator is $\{A, B\} = AB + BA$.
:::

:::{admonition} Derivation: Boson Commutation Relations from Definitions
:class: dropdown note

We prove $[b_\alpha, b^\dagger_\beta] = \delta_{\alpha\beta}$ using the action on Fock states.

**Case 1: $\alpha = \beta$**

$$


[b_\alpha, b^\dagger_\alpha] |n_1, \ldots, n_\alpha, \ldots\rangle = b_\alpha b^\dagger_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle - b^\dagger_\alpha b_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle

$$


Apply $b^\dagger_\alpha$ first:

$$


b^\dagger_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = \sqrt{n_\alpha + 1} |n_1, \ldots, n_\alpha + 1, \ldots\rangle

$$


Then $b_\alpha$:

$$


b_\alpha \sqrt{n_\alpha + 1} |n_1, \ldots, n_\alpha + 1, \ldots\rangle = \sqrt{n_\alpha + 1} \cdot \sqrt{n_\alpha + 1} |n_1, \ldots, n_\alpha, \ldots\rangle = (n_\alpha + 1) |n_1, \ldots, n_\alpha, \ldots\rangle

$$


Now apply $b_\alpha$ first, then $b^\dagger_\alpha$:

$$


b_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = \sqrt{n_\alpha} |n_1, \ldots, n_\alpha - 1, \ldots\rangle

$$


$$

b^\dagger_\alpha \sqrt{n_\alpha} |n_1, \ldots, n_\alpha - 1, \ldots\rangle = \sqrt{n_\alpha} \cdot \sqrt{n_\alpha} |n_1, \ldots, n_\alpha, \ldots\rangle = n_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle

$$


The commutator:

$$


[b_\alpha, b^\dagger_\alpha] = (n_\alpha + 1) - n_\alpha = 1 \quad ✓

$$


**Case 2: $\alpha \neq \beta$**

Since $b_\alpha$ and $b^\dagger_\beta$ act on different states, they commute:

$$


[b_\alpha, b^\dagger_\beta] = 0 \quad ✓

$$


Similarly, $[b_\alpha, b_\beta] = 0$ and $[b^\dagger_\alpha, b^\dagger_\beta] = 0$ follow from the fact that creation and annihilation operators on **different states** do not interfere.
:::

::{admonition} Derivation: Pauli Exclusion from Anticommutation Algebra
:class: dropdown note

From $\{c^\dagger_\alpha, c^\dagger_\alpha\} = 0$:

$$


c^\dagger_\alpha c^\dagger_\alpha + c^\dagger_\alpha c^\dagger_\alpha = 0 \quad \Rightarrow \quad c^\dagger_\alpha c^\dagger_\alpha = 0

$$


This means applying $c^\dagger_\alpha$ twice to any state gives **zero**—impossible to put two fermions in the same state! This is the Pauli exclusion principle encoded in the algebra.
:::

### Physical Meaning

The choice between commutators (bosons) and anticommutators (fermions) is **not arbitrary**; it is **dictated** by the symmetry of the wavefunction:

- **Commuting operators**: particle exchange does not change order → symmetric (bosonic) states.
- **Anticommuting operators**: particle exchange picks up a minus sign → antisymmetric (fermionic) states.
:::{admonition} Derivation: Fermion Anticommutation Relations
:class: dropdown note

We prove $\{c_\alpha, c^\dagger_\beta\} = \delta_{\alpha\beta}$ using the action on Fock states (with $n_i \in \{0, 1\}$).

**Case 1: $\alpha = \beta$**

Consider a state $|n_1, \ldots, n_\alpha, \ldots\rangle$ where $n_\alpha \in \{0, 1\}$.

**Subcase 1a: $n_\alpha = 0$** (state is empty)

$$


c^\dagger_\alpha |\ldots, 0, \ldots\rangle = |\ldots, 1, \ldots\rangle

$$


$$

c_\alpha |\ldots, 1, \ldots\rangle = |\ldots, 0, \ldots\rangle

$$


So: $c_\alpha c^\dagger_\alpha |\ldots, 0, \ldots\rangle = |\ldots, 0, \ldots\rangle$

And: $c^\dagger_\alpha c_\alpha |\ldots, 0, \ldots\rangle = 0$ (cannot annihilate from empty state)

The anticommutator:

$$


\{c_\alpha, c^\dagger_\alpha\} |\ldots, 0, \ldots\rangle = (1 + 0) |\ldots, 0, \ldots\rangle = |\ldots, 0, \ldots\rangle \quad ✓

$$


**Subcase 1b: $n_\alpha = 1$** (state is occupied)

$$


c^\dagger_\alpha |\ldots, 1, \ldots\rangle = 0

$$


$$

c_\alpha |\ldots, 1, \ldots\rangle = |\ldots, 0, \ldots\rangle

$$


So: $c_\alpha c^\dagger_\alpha |\ldots, 1, \ldots\rangle = 0$

And: $c^\dagger_\alpha c_\alpha |\ldots, 1, \ldots\rangle = c^\dagger_\alpha |\ldots, 0, \ldots\rangle = |\ldots, 1, \ldots\rangle$

The anticommutator:

$$


\{c_\alpha, c^\dagger_\alpha\} |\ldots, 1, \ldots\rangle = (0 + 1) |\ldots, 1, \ldots\rangle = |\ldots, 1, \ldots\rangle \quad ✓

$$


**Case 2: $\alpha \neq \beta$**

Since $c_\alpha$ and $c^\dagger_\beta$ act on different states:

$$


\{c_\alpha, c^\dagger_\beta\} = 0 \quad ✓

$$


The phases ($(-1)^P$) on different states are independent and cancel in the anticommutation relation.
:::
## The Number Operator

### Definition

The **number operator** (or **occupation number operator**) for state $\alpha$ is:

$$


\hat{n}_\alpha = b^\dagger_\alpha b_\alpha \quad \text{(bosons)}

$$


$$

\hat{n}_\alpha = c^\dagger_\alpha c_\alpha \quad \text{(fermions)}

$$ (eq-number-operator)

The same form applies to both! The difference lies in the algebra of the operators.

### Eigenvalues and Eigenstates

The action on Fock states:

$$


\hat{n}_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle = n_\alpha |n_1, \ldots, n_\alpha, \ldots\rangle

$$ (eq-number-eigenvalue)

Fock states are **eigenstates** of number operators, with eigenvalues equal to the occupation numbers.

### Total Particle Number

The **total particle number operator**:

$$


\hat{N} = \sum_\alpha \hat{n}_\alpha = \sum_\alpha b^\dagger_\alpha b_\alpha

$$ (eq-total-number)

Eigenvalue on state $|n_1, n_2, \ldots\rangle$:

$$


\hat{N} |n_1, n_2, \ldots\rangle = \left(\sum_\alpha n_\alpha\right) |n_1, n_2, \ldots\rangle

$$


This allows systems with a fixed particle number or (in the full Fock space) variable particle number.
## Constructing States with Operators

### Building from the Vacuum

Any Fock state can be built by repeatedly applying creation operators to the vacuum:

$$


|n_1, n_2, \ldots, n_D\rangle = \frac{(b^\dagger_1)^{n_1} (b^\dagger_2)^{n_2} \cdots (b^\dagger_D)^{n_D}}{\sqrt{n_1! n_2! \cdots n_D!}} |0\rangle \quad \text{(bosons)}

$$ (eq-boson-state-construction-second-quantization)

For fermions (where $n_\alpha \in \{0, 1\}$):

$$


|n_1, n_2, \ldots, n_D\rangle = (c^\dagger_1)^{n_1} (c^\dagger_2)^{n_2} \cdots (c^\dagger_D)^{n_D} |0\rangle \quad \text{(fermions)}

$$ (eq-fermion-state-construction)

where the exponents are either 0 or 1.

:::{admonition} Example: Two Bosons
:class: example


$$


|2, 0, 0, \ldots\rangle = \frac{(b^\dagger_1)^2}{\sqrt{2}} |0\rangle

$$


Check: $b_1 |2, 0, \ldots\rangle = \sqrt{2} |1, 0, \ldots\rangle$.

### Example: Two Fermions in Different States

$$


|1, 1, 0, \ldots\rangle = c^\dagger_1 c^\dagger_2 |0\rangle

$$


If we try $|1, 1, 0, \ldots\rangle = c^\dagger_1 c^\dagger_1 |0\rangle$, anticommutation gives $(c^\dagger_1)^2 = 0$, so the state is **zero**—impossible to put two fermions in state 1.
:::
## Field Operators (Optional: Introduction)

In many-body physics and quantum field theory, **field operators** generalize creation/annihilation operators to continuous space:

$$


\hat{\psi}(\boldsymbol{r}) = \sum_\alpha \phi_\alpha(\boldsymbol{r}) b_\alpha \quad \text{(bosons)}

$$


$$

\hat{\psi}^\dagger(\boldsymbol{r}) = \sum_\alpha \phi^*_\alpha(\boldsymbol{r}) b^\dagger_\alpha \quad \text{(bosons)}

$$ (eq-field-operator)

where $\phi_\alpha(\boldsymbol{r})$ are single-particle wavefunction basis functions.

These field operators:
- Create or destroy a particle at a specific **position** $\boldsymbol{r}$ (spatial localization).
- Still satisfy commutation (bosons) or anticommutation (fermions) relations.
- Form the foundation of quantum field theory and are essential for discussing particle interactions in a many-body context.
## Second Quantization of a Hamiltonian

### Single-Particle Hamiltonian

A first-quantized Hamiltonian for non-interacting particles:

$$


H_0^{(1)} = \sum_{i=1}^N h(\boldsymbol{r}_i)

$$


becomes, in second quantization:

$$


\hat{H}_0 = \sum_\alpha \epsilon_\alpha b^\dagger_\alpha b_\alpha

$$ (eq-secondq-single-particle)

where $\epsilon_\alpha$ are single-particle energies (eigenvalues of $h$).

### Two-Body Interaction

An interaction between pairs of particles:

$$


H_{int}^{(1)} = \frac{1}{2} \sum_{i \neq j} V(\boldsymbol{r}_i - \boldsymbol{r}_j)

$$


becomes:

$$


\hat{H}_{int} = \frac{1}{2} \sum_{\alpha,\beta,\gamma,\delta} V_{\alpha\beta\gamma\delta} b^\dagger_\alpha b^\dagger_\beta b_\gamma b_\delta

$$ (eq-secondq-interaction)

where $V_{\alpha\beta\gamma\delta} = \langle \alpha\beta | V | \gamma\delta \rangle$ are matrix elements.

### Advantages of Second Quantization

1. **Automatic symmetry**: No need to symmetrize/antisymmetrize manually.
2. **Variable particle number**: Easy to handle states with different $N$.
3. **Compact notation**: Expressions are concise and amenable to perturbation theory.
4. **Natural for field theory**: Connects directly to quantum field theory.
## Summary

- **Second quantization** replaces "which particle in which state" with "how many particles in each state?"
- **Fock states** labeled by occupation numbers $\{n_\alpha\}$ form a complete, orthonormal basis.
- **Creation operator** $b^\dagger_\alpha$ (or $c^\dagger_\alpha$) adds a particle to state $\alpha$.
- **Annihilation operator** $b_\alpha$ (or $c_\alpha$) removes a particle from state $\alpha$.
- **Bosons**: occupation unlimited ($n_\alpha \in \{0, 1, 2, \ldots\}$), commutation relations $[b_\alpha, b^\dagger_\beta] = \delta_{\alpha\beta}$.
- **Fermions**: occupation $n_\alpha \in \{0, 1\}$, anticommutation relations $\{c_\alpha, c^\dagger_\beta\} = \delta_{\alpha\beta}$.
- **Anticommutation encodes Pauli exclusion**: $\{c^\dagger_\alpha, c^\dagger_\alpha\} = 0$ forbids double occupation.
- **Number operator** $\hat{n}_\alpha = b^\dagger_\alpha b_\alpha$ counts particles in state $\alpha$.
