## Lecture Notes

The Aharonov-Bohm effect shatters the classical intuition that electromagnetic effects require local presence of fields. A quantum particle can acquire observable phase shifts from vector potentials in regions where the magnetic field vanishes entirely, revealing that quantum mechanics couples to potentials, not fields. This demonstrates a fundamental non-locality: the quantum world is sensitive to long-range, gauge-dependent objects, and only gauge-invariant combinations (like loop integrals of the potential) have physical meaning.
### The Experimental Setup

Consider a very long solenoid of radius $R$ carrying a magnetic field $\boldsymbol{B} = B\hat{z}$ entirely contained within. The solenoid is so thin that its length is much greater than its radius. **Outside the solenoid**, the magnetic field vanishes: $\boldsymbol{B}(r > R) = 0$.

In classical electromagnetism, particles traveling in the field-free region experience no Lorentz force and move in straight lines. The magnetic field is "hidden" inside the solenoid and has no influence outside.

In quantum mechanics, the situation is dramatically different. Although $\boldsymbol{B} = 0$ outside, the **vector potential** $\boldsymbol{A}$ is non-zero. This creates observable quantum interference effects—the **Aharonov-Bohm (AB) effect**.

:::{admonition} The Key Insight
:class: important

The Aharonov-Bohm effect demonstrates that the vector potential $\boldsymbol{A}$ is **physically fundamental**, not just a mathematical auxiliary field. In quantum mechanics, charged particles couple directly to the potentials, not just to the fields.
:::

### Vector Potential of the Solenoid

For a solenoid along the $z$-axis with internal field $\boldsymbol{B} = B\hat{z}$ and radius $R$:

**Inside** ($r < R$):

$$

\boldsymbol{A}_\text{in} = \frac{B}{2}\boldsymbol{\hat{z}} \times \boldsymbol{r} = \frac{Br}{2}\hat{\theta}

$$

where $(r, \theta, z)$ are cylindrical coordinates.

**Outside** ($r > R$):

$$

\boldsymbol{A}_\text{out} = \frac{BR^2}{2r}\hat{\theta}

$$

The key feature: the vector potential has an **azimuthal component** that doesn't vanish outside. At $r = R$ (the solenoid boundary), both expressions agree: $A_\theta = BR/2$, ensuring continuity.

The magnitude of the potential falls as $1/r$ for large $r$, much slower than the field (which falls exponentially from the boundary).

### Circulation of Vector Potential

The total magnetic flux through the solenoid is:

$$

\Phi_B = \int_{\text{inside}} \boldsymbol{B} \cdot d\boldsymbol{S} = B \cdot \pi R^2

$$

A crucial result follows from **Stokes' theorem**. When we integrate the vector potential around any closed path that encircles the solenoid:

$$

\oint \boldsymbol{A} \cdot d\boldsymbol{l} = \int (\nabla \times \boldsymbol{A}) \cdot d\boldsymbol{S} = \int \boldsymbol{B} \cdot d\boldsymbol{S} = \Phi_B

$$

This is a **remarkable result**: the line integral of the vector potential around the solenoid equals the total enclosed magnetic flux. This means that although the field vanishes outside, the circulation of the potential encodes the complete flux information.

:::{admonition} Derivation: Circulation from Azimuthal Symmetry
:class: dropdown note


For a circular path of radius $r > R$ (outside the solenoid) in the azimuthal direction:

$$

\oint \boldsymbol{A} \cdot d\boldsymbol{l} = \int_0^{2\pi} A_\theta(r) \, r \, d\theta

$$

Outside the solenoid, $A_\theta = \frac{BR^2}{2r}$, so:

$$

\oint \boldsymbol{A} \cdot d\boldsymbol{l} = \int_0^{2\pi} \frac{BR^2}{2r} \, r \, d\theta = \int_0^{2\pi} \frac{BR^2}{2} d\theta = \frac{BR^2}{2} \cdot 2\pi = \pi B R^2 = \Phi_B

$$

The result is independent of the integration path's radius and shape—it depends only on whether the path encloses the flux.

:::

### Phase Accumulation Along Closed Path

In quantum mechanics, when a charged particle travels along a path $C$ in the presence of a vector potential, it acquires a **phase factor**. For a closed path encircling the solenoid, the observable phase shift is:

$$

\phi_{\text{AB}} = \frac{e\Phi_B}{\hbar c}

$$

This is the **Aharonov-Bohm phase**. Remarkably, it depends **only on the enclosed flux**, not on the detailed field distribution or the path's shape. For typical experimental values (magnetic flux $\Phi_B \sim 10^{-15}$ Wb, de Broglie wavelength $\lambda \sim 0.1$ nm), this phase is typically several radians—easily measurable through interference.

:::{admonition} Derivation: Phase from Path Integral
:class: dropdown note


When a charged particle travels along path $C$, the wavefunction acquires phase:

$$

\psi(\text{end}) = e^{i\phi_C} \psi(\text{start})

$$

where the phase is:

$$

\phi_C = \frac{q}{\hbar c}\int_C \boldsymbol{A} \cdot d\boldsymbol{l}

$$

For a closed path encircling the solenoid, this becomes:

$$

\phi_{\text{AB}} = \frac{q\Phi_B}{\hbar c}

$$

Using $q = e$ for an electron:

$$

\phi_{\text{AB}} = \frac{e\Phi_B}{\hbar c}

$$

:::{admonition} Units and Conventions
:class: note

In SI units, $\phi = \frac{e}{\hbar c}\int \boldsymbol{A} \cdot d\boldsymbol{l}$. In Gaussian units (often used in older texts), $\phi = \frac{2e}{hc}\int \boldsymbol{A} \cdot d\boldsymbol{l}$. The factors differ between conventions; check your reference carefully.

:::

:::

### Double-Slit Interference Pattern Shift

Consider electrons passing through two slits, with a solenoid positioned so that one slit is above it and one below. Electrons can take two paths:

**Path 1** (above the solenoid): Encircles the solenoid, accumulating phase $\phi_1 = \phi_0 + \phi_{\text{AB}}/2$ (half the total flux on one side)

**Path 2** (below the solenoid): Also encircles, but in opposite sense: $\phi_2 = \phi_0 - \phi_{\text{AB}}/2$

The **phase difference** between the two paths is:

$$

\Delta\phi = \phi_1 - \phi_2 = \frac{e\Phi_B}{\hbar c}

$$

In the interference region, the intensity is:

$$

I(x) = |\psi_1 + \psi_2|^2 = |\psi_1|^2 + |\psi_2|^2 + 2|\psi_1||\psi_2|\cos(\Delta\phi)

$$

The interference pattern (bright and dark fringes) **shifts** by an amount proportional to the change in $\Delta\phi$. Varying the solenoid current (and thus $\Phi_B$) causes the entire pattern to shift while the electron paths remain unchanged (since $\boldsymbol{B} = 0$ outside).

**This pattern shift is the observable signature of the AB effect.** Remarkably, it depends only on $\Phi_B$, not on the detailed field distribution inside the solenoid or its size.

### Non-Locality and Path Independence

A crucial aspect of the AB effect is its **non-local** character:

1. **The particle never enters the region of strong field** (where $\boldsymbol{B} \neq 0$), yet it experiences an effect.

2. **The phase depends only on the global flux**, not on the local field strength at the particle's location.

3. **The effect persists even if the solenoid is infinitesimally thin** but carries arbitrarily large flux (the thin solenoid limit).

This demonstrates a fundamental principle: **in quantum mechanics, particles couple to potentials, which are intrinsically non-local objects**. The wavefunction "feels" the long-range vector potential, accumulating phase far from the source of the field.

### Experimental Observation

The Aharonov-Bohm effect was experimentally verified by **Tonomura et al. (1986)** using electron holograms. They used:

- A thin, insulated toroidal magnet to create controlled magnetic flux
- Electron beam splitters and recombination optics
- Electron holography to measure the phase shift with nanometer-scale spatial resolution

The experiment directly mapped the phase shift as a function of position, definitively proving:
1. The vector potential has physical significance in quantum mechanics
2. The phase shift depends only on the enclosed flux
3. The effect is gauge-invariant (independent of gauge choice for $\boldsymbol{A}$)

### Relation to Berry Phase

The Aharonov-Bohm phase can be understood as a **Berry phase** acquired as the particle's momentum adiabatically cycles through parameter space. The Berry connection in parameter space encodes the vector potential in real space, showing a deep connection between geometric phases and gauge fields.

### Summary

- **AB effect**: Charged particles acquire phase $\phi = \frac{q\Phi_B}{\hbar c}$ when traveling around enclosed magnetic flux, even in field-free regions.
- **Non-locality**: The effect demonstrates that quantum mechanics is fundamentally non-local—particles couple to long-range potentials, not just local fields.
- **Potentials fundamental**: Vector potential $\boldsymbol{A}$ directly affects wavefunction evolution; it's not just an auxiliary field.
- **Gauge potentials observable**: Although $\boldsymbol{A}$ is gauge-dependent, **line integrals** $\oint \boldsymbol{A} \cdot d\boldsymbol{l}$ are observable through interference.
- **Pattern shift**: Interference fringes shift by amount proportional to $\Phi_B$; shift depends only on total flux, not field details.
- **Experimental confirmation**: Tonomura's electron holography definitively verified the effect (1986).
- **Topological origin**: The AB phase is topological—depends only on the winding number of the flux, insensitive to deformations of the path.
