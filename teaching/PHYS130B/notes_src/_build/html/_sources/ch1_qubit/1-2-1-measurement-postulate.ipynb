## Lecture Notes



Quantum measurement is not a passive observation—it is a transformation that collapses the state to one of the observable's eigenstates with probability given by the Born rule. This section formalizes measurement as a fundamental postulate of quantum mechanics and explores the profound implications of measurement-induced collapse for how we understand quantum systems.

### The Three Axioms of Measurement

Quantum mechanics makes precise predictions about measurement outcomes through the **measurement postulate**. This postulate has three parts:
#### Axiom 1: Possible Outcomes

When we measure an observable $\hat{O}$ (a Hermitian operator), the **only possible outcomes** are the **eigenvalues** of $\hat{O}$.

:::{admonition} Example: Spin-z Eigenvalues
:class: example

For $\hat{\sigma}^z$, the eigenvalues are $\pm 1$ with eigenstates $|0\rangle$ and $|1\rangle$.

:::
#### Axiom 2: The Born Rule (Probabilities)

Suppose the system is in state $|\psi\rangle$, and we measure observable $\hat{O}$. Let $\{|\phi_i\rangle\}$ be the **complete set of orthonormal eigenvectors** of $\hat{O}$, with corresponding eigenvalues $\lambda_i$:

$$


\hat{O} |\phi_i\rangle = \lambda_i |\phi_i\rangle

$$


The **probability** of obtaining outcome $\lambda_i$ is

$$


P_i = |\langle \phi_i | \psi \rangle|^2

$$ (eq-born-rule)

This is the **Born rule**: the probability is the **square of the amplitude** of the overlap between the state and the eigenstate.

**Normalization**: The probabilities sum to 1 because $\{|\phi_i\rangle\}$ form a complete orthonormal basis.

:::{admonition} Verification: Probability Normalization
:class: dropdown note

$$


\sum_i P_i = \sum_i |\langle \phi_i | \psi \rangle|^2 = 1

$$


This follows from the completeness relation $\sum_i |\phi_i\rangle\langle\phi_i| = I$ applied to the normalized state $|\psi\rangle$.
:::


:::{admonition} Discussion
:class: dropdown warning

**Why do we square the amplitude instead of using linear probability?**

Consider the double-slit experiment: a particle in state $|\psi\rangle = \frac{1}{\sqrt{2}}(|\text{slit 1}\rangle + |\text{slit 2}\rangle)$ passes through both slits. 

Without the Born rule (linear version): $P = |c_1| + |c_2| = \frac{1}{\sqrt{2}} + \frac{1}{\sqrt{2}} = \sqrt{2}$ — this exceeds 1 and violates probability! 

With the Born rule (squared): $P = |c_1|^2 + |c_2|^2 + 2\text{Re}(c_1^* c_2) = \frac{1}{2} + \frac{1}{2} + \text{interference term}$. 

**Your task**: Show that the interference term $2\text{Re}(c_1^* c_2)$ is nonzero for this equal superposition, and explain why the Born rule's squared amplitudes are essential to correctly predict quantum interference. What happens to the interference if you insert a which-path detector that projects onto a definite slit?

:::
#### Axiom 3: State Collapse

Immediately after the measurement returns outcome $\lambda_i$, the state **collapses** (or **projects**) to the corresponding eigenstate:

$$


|\psi\rangle_{\text{after}} = |\phi_i\rangle

$$ (eq-state-collapse)

This is sometimes called **wavefunction collapse** or the **projection postulate**. It is irreversible: the old state information is lost unless we knew the outcome beforehand.

:::{admonition} Key Insight: Measurement is Destructive
:class: important

Measurement is fundamentally different from observation in classical mechanics. In classical physics, we can look at a system without changing it. In quantum mechanics, measurement **irreversibly changes the state**. This is not due to experimental imperfection—it is built into the theory.

After measurement, all information about the superposition is gone. The state is now in a definite eigenstate, with complete certainty about that eigenvalue.
:::

Consider a qubit in the state $|\psi\rangle = \frac{1}{\sqrt{3}}|0\rangle + \sqrt{\frac{2}{3}}|1\rangle$ and we measure $\hat{\sigma}^z$.

:::{admonition} Example: Probabilities and State Collapse
:class: dropdown note

The eigenvalues are $\lambda_0 = +1$ (eigenstate $|0\rangle$) and $\lambda_1 = -1$ (eigenstate $|1\rangle$).

**Probability of outcome +1**:

$$


P_+ = |\langle 0 | \psi \rangle|^2 = \left|\frac{1}{\sqrt{3}}\right|^2 = \frac{1}{3}

$$


**Probability of outcome -1**:

$$


P_- = |\langle 1 | \psi \rangle|^2 = \left|\sqrt{\frac{2}{3}}\right|^2 = \frac{2}{3}

$$


**Verification**: $P_+ + P_- = 1/3 + 2/3 = 1$ ✓
:::

**State after measurement**:
- If we observe outcome $+1$, the state collapses to $|0\rangle$.
- If we observe outcome $-1$, the state collapses to $|1\rangle$.

After the measurement, all memory of the original superposition is erased.

Consider a qubit in state $|0\rangle$ and we measure $\hat{\sigma}^x$.

:::{admonition} Example: Complementarity—Different Bases
:class: dropdown note

The eigenvalues of $\hat{\sigma}^x$ are $\pm 1$ with eigenstates:

$$


|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \quad
|-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)

$$


**Probability of outcome +1**:

$$


P_+ = |\langle + | 0 \rangle|^2 = \left|\frac{1}{\sqrt{2}}\right|^2 = \frac{1}{2}

$$


**Probability of outcome -1**:

$$


P_- = |\langle - | 0 \rangle|^2 = \left|\frac{1}{\sqrt{2}}\right|^2 = \frac{1}{2}

$$


:::

**Interpretation**: Even though we prepared spin-up along z with certainty, measurement along x gives 50-50 outcomes. The z-basis state is a superposition in the x-basis:

$$


|0\rangle = \frac{1}{\sqrt{2}}|+\rangle + \frac{1}{\sqrt{2}}|-\rangle

$$

:class: dropdown warning

**The Quantum Zeno Effect: how does repeated measurement slow decay?**

For a two-level system with decay rate $\Gamma$, consider a rapidly oscillating Rabi frequency $\Omega$ and $N$ measurements over total time $T$. The survival probability (probability of no decay) after $N$ equally-spaced measurements is approximately

$$


P_N(T) = \cos^{2N}\left(\frac{\Omega T}{2N}\right).

$$


**Your task**:

(a) In the limit $N \to \infty$ (infinitely frequent measurements), show that $P_N(T) \to 1$. That is, infinitely rapid projections freeze the system. (Hint: use $\lim_{n \to \infty} \cos(x/n)^n = 1$ for small $x$.)

(b) At what measurement frequency (number of measurements per unit time) does the Zeno effect become observable? Estimate the measurement rate needed to resolve decay in your favorite atomic system.

(c) This seems to contradict the idea that "collapse is just a bookkeeping update"—rapid measurements are doing *real physical work* by preventing evolution. Does the Zeno effect resolve the debate about whether collapse is physical? Or does it only tell us that *collapse (whether real or not) has observable consequences*?

:::
This is **quantum complementarity**: non-commuting observables cannot be simultaneously definite.

### Expectation Values from the Born Rule

The **expectation value** (average outcome over many measurements) is

$$


\langle \hat{O} \rangle = \sum_i \lambda_i P_i = \sum_i \lambda_i |\langle \phi_i | \psi \rangle|^2

$$ (eq-expectation-from-probabilities)

This can be rewritten (using completeness of eigenbases) as

$$


\langle \hat{O} \rangle = \langle \psi | \hat{O} | \psi \rangle

$$ (eq-expectation-operator)

For the state $|\psi\rangle = \frac{1}{\sqrt{3}}|0\rangle + \sqrt{\frac{2}{3}}|1\rangle$ and $\hat{\sigma}^z$:

:::{admonition} Verification: Expectation Value
:class: dropdown note

$$


\langle \hat{\sigma}^z \rangle = (+1) \cdot \frac{1}{3} + (-1) \cdot \frac{2}{3} = \frac{1}{3} - \frac{2}{3} = -\frac{1}{3}

$$


We can verify with the operator formula:

$$


\langle \psi | \hat{\sigma}^z | \psi \rangle = \left(\frac{1}{\sqrt{3}}\langle 0| + \sqrt{\frac{2}{3}}\langle 1|\right) \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \left(\frac{1}{\sqrt{3}}|0\rangle + \sqrt{\frac{2}{3}}|1\rangle\right)

$$


$$

= \left(\frac{1}{\sqrt{3}}\langle 0| + \sqrt{\frac{2}{3}}\langle 1|\right) \left(\frac{1}{\sqrt{3}}|0\rangle - \sqrt{\frac{2}{3}}|1\rangle\right) = \frac{1}{3} - \frac{2}{3} = -\frac{1}{3} \quad \checkmark

$$


:::

### Variance and Uncertainty

The **variance** of a measurement is

$$


\Delta^2 \hat{O} = \langle \hat{O}^2 \rangle - \langle \hat{O} \rangle^2 = \langle \psi | \hat{O}^2 | \psi \rangle - (\langle \psi | \hat{O} | \psi \rangle)^2

$$ (eq-variance)

The **standard deviation** is $\Delta \hat{O} = \sqrt{\Delta^2 \hat{O}}$.

**Physical interpretation**: $\Delta \hat{O}$ measures the spread in measurement outcomes. If the state is an eigenstate of $\hat{O}$, then $\Delta \hat{O} = 0$ (all measurements give the same result).

For the state $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ and observable $\hat{\sigma}^z$:

:::{admonition} Calculation: Variance for Superposition State
:class: dropdown note

$$


\langle \hat{\sigma}^z \rangle = 0 \quad \text{(equal superposition)}

$$


$$

\langle (\hat{\sigma}^z)^2 \rangle = \langle + | I | + \rangle = 1 \quad \text{(since } (\hat{\sigma}^z)^2 = I\text{)}

$$


$$

\Delta^2 \hat{\sigma}^z = 1 - 0 = 1 \implies \Delta \hat{\sigma}^z = 1

$$


:::

This makes sense: we get $\pm 1$ with equal probability, so outcomes fluctuate maximally around the mean of 0.

### Projective Measurement

The **projection operator** onto eigenstate $|\phi_i\rangle$ is

$$


P_i = |\phi_i\rangle\langle \phi_i|

$$ (eq-projection-operator)

**Properties**:
- $P_i^2 = P_i$ (idempotence): applying twice is the same as applying once.
- $P_i^\dagger = P_i$ (Hermiticity).
- $\sum_i P_i = I$ (completeness).

**Measurement probability** (alternative formula):

$$


P_i = \langle \psi | P_i | \psi \rangle = |\langle \phi_i | \psi \rangle|^2

$$


**State collapse** (alternative formula):

$$


|\psi\rangle_{\text{after}} = \frac{P_i |\psi\rangle}{\sqrt{\langle \psi | P_i | \psi \rangle}} = \frac{|\phi_i\rangle\langle \phi_i | \psi \rangle}{|\langle \phi_i | \psi \rangle|}

$$


### Summary

- The **measurement postulate** has three parts: (1) outcomes are eigenvalues, (2) probabilities follow the Born rule $P_i = |\langle\phi_i|\psi\rangle|^2$, (3) the state collapses to the corresponding eigenstate.
- **Measurement is irreversible**: information about superpositions is destroyed.
- **Expectation values** are averages over many measurements and can be computed as $\langle\hat{O}\rangle = \langle\psi|\hat{O}|\psi\rangle$.
- **Variance** ($\Delta \hat{O}$) quantifies the spread in measurement outcomes; zero for eigenstates, maximal for superpositions of eigenstates with equal eigenvalues.
- **Projective measurement** provides a formal framework: measurement is projection onto an eigenspace, with probability given by the norm squared of the projected state.
- The full theory of measurement, including the generalized framework beyond projective measurements, is developed in **§6.3 of Chapter 6**.
