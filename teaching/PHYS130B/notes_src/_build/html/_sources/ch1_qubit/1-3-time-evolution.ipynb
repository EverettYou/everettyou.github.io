# 1.3 Time Evolution

## Overview

Quantum systems evolve in time according to the Schrödinger equation, the fundamental law of quantum dynamics. This section explores how quantum states develop, why the evolution must preserve information (unitarity), and two complementary perspectives—Schrödinger and Heisenberg pictures—that reveal deep connections to symmetry and conservation laws.

**In quantum computing:** Quantum gates are realized by engineering precise time evolution under tailored Hamiltonians. A $\pi/2$-pulse creates superposition; a $\pi$-pulse flips qubits. Understanding time evolution is essential for designing quantum circuits and controlling qubits.

**In atomic and molecular physics:** Electronic transitions, spectroscopy, magnetic resonance (NMR, ESR), and chemical reactions all involve time-dependent evolution. The Larmor precession of spins in magnetic fields and Rabi oscillations under resonant driving are central to manipulating quantum systems.

**In fundamental physics:** The Schrödinger equation governs all phenomena from atoms to stars. Symmetries and their generators (conservation laws) emerge naturally from the Heisenberg picture and extend to the gauge symmetries of the Standard Model (U(1), SU(2), SU(3)).


### Key Concepts

- **Schrödinger equation**: $\mathrm{i}\hbar \frac{\partial\vert\psi\rangle}{\partial t} = \hat{H}\vert\psi\rangle$, the master equation connecting state evolution to the Hamiltonian (energy operator).

- **Time-evolution operator**: $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$ is unitary and directly solves the Schrödinger equation. Unitarity ensures normalization and probability conservation.

- **Stationary states**: Energy eigenstates $\vert E_n\rangle$ are stationary—they evolve only by an unobservable global phase $\mathrm{e}^{-\mathrm{i}E_n t/\hbar}$. Physical quantities remain constant.

- **Superposition dynamics**: Superpositions of energy eigenstates exhibit quantum interference due to different phase evolution rates. This produces oscillatory behavior and is the origin of Rabi oscillations.

- **Larmor precession**: A spin in a magnetic field precesses around the field direction at the **Larmor frequency** $\omega_0 = \gamma B_0$. The Bloch vector rotates on the Bloch sphere.

- **Rabi oscillations**: When a two-level system is driven by an oscillating field resonant with the transition, population oscillates periodically between levels. The **Rabi frequency** $\Omega$ determines the oscillation rate.

- **Heisenberg picture**: Complementary formulation where operators evolve and states are static. The **Heisenberg equation** $\frac{\mathrm{d}\hat{O}}{\mathrm{d}t} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{O}]$ governs operator dynamics.

- **Conservation laws**: An observable $\hat{A}$ is conserved ($\frac{\mathrm{d}}{\mathrm{d}t}\langle \hat{A} \rangle = 0$) if and only if it commutes with the Hamiltonian: $[\hat{A}, \hat{H}] = 0$. This connects symmetries to conserved quantities.

- **Ehrenfest theorem**: Expectation values follow classical equations of motion, revealing the quantum-to-classical correspondence.

### Learning Objectives

By the end of this section, you should be able to:

- State and solve the time-dependent Schrödinger equation for simple Hamiltonians (especially two-level systems)
- Understand unitarity, why it is essential, and what it guarantees (information preservation, probability conservation)
- Recognize and explain stationary states (energy eigenstates) and their role in quantum dynamics
- Compute time evolution of superpositions and predict interference and oscillation effects
- Apply the Rabi formula to analyze and design quantum gates ($\pi/2$-pulse, $\pi$-pulse, resonant driving)
- Explain Larmor precession: how a spin's Bloch vector rotates in a magnetic field at the Larmor frequency
- Distinguish between Schrödinger and Heisenberg pictures and use both appropriately
- Apply the Heisenberg equation to find conserved quantities and operator time-dependence
- Connect symmetries to conservation laws: if $[\hat{G}, \hat{H}] = 0$, then $\hat{G}$ is conserved
- Use the Ehrenfest theorem to bridge quantum and classical dynamics

## Projects

### Project: Pulse Shaping and GRAPE Optimization for Quantum Gate Design

**Objective:** Develop optimal control techniques (GRAPE: Gradient Ascent Pulse Engineering) to design smooth, robust pulse sequences that implement target quantum gates under realistic Hamiltonian constraints and decoherence.

**Background:** Ideal quantum gates assume infinite control bandwidth and zero error. In practice, control pulses must be smooth, band-limited, and robust to parameter variations. GRAPE and similar optimal control methods are at the frontier of quantum device engineering. They solve the inverse problem: given a target unitary, find the control fields $\Omega(t)$ that generate it under the true microscopic Hamiltonian, accounting for decoherence and driving constraints. This project deepens understanding of time evolution (1.3) by solving it in the presence of realistic imperfections.

**Suggested Approach:**
- **Literature survey**: Study optimal control in quantum systems. Read "GRAPE: A Practical Approach to Quantum Optimal Control" or similar. Understand how gradient descent optimizes fidelity with respect to control parameters.
- **Hamiltonian model**: Work with a spin in a magnetic field: $\hat{H} = \frac{\omega_0}{2}\hat{\sigma}^z + \Omega(t)\hat{\sigma}^x$ (resonant drive, rotating frame). Include a decoherence term: dephasing rate $\gamma_\phi$, spontaneous emission $\gamma_1$.
- **Optimization problem**: Minimize the infidelity $\mathcal{I} = 1 - \frac{1}{2}|\mathrm{Tr}(U_{\text{target}}^\dagger U_{\text{achieved}})|$ by choosing the pulse shape $\Omega(t)$.
- **Implementation**: Code a GRAPE solver using gradient-based optimization (scipy.optimize). Parametrize the pulse as a sum of basis functions (e.g., Fourier components or piecewise constant intervals). Solve the Liouville-von Neumann equation numerically with decoherence.
- **Benchmarks**: Design pulses for: (a) $\pi/2$ rotation (Hadamard-like), (b) controlled NOT gate (using 2-qubit coupling), (c) resilience to pulse amplitude errors.
- **Analysis**: Compare GRAPE-optimized pulses vs. simple analytical pulses. Show how decoherence time $T_2$ sets the shortest achievable gate time. Discuss scalability.

**Expected Deliverable:** Research report (6-8 pages) with: optimal control theory overview, GRAPE algorithm derivation, numerical results for 1-2 qubit gates, fidelity vs. gate time trade-offs, robustness to parameter variations, and discussion of experimental implementation on current quantum devices.

**Key References:** Khaneja et al. (2005) "Optimal control of coupled spin dynamics"; "Optimal Control Methods for Quantum Systems"; recent GRAPE implementations on superconducting qubits (IBM, Google).
