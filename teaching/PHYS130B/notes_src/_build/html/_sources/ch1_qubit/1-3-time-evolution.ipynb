# 1.3 Time Evolution

## Overview

This section explores how quantum states evolve in time—the central dynamical law of quantum mechanics. The Schrödinger equation governs all quantum processes, from electronic transitions in atoms to quantum gate operations in computers.

**In quantum computing:** Quantum gates are realized by controlled evolution under tailored Hamiltonians. A $\pi/2$-rotation (Hadamard gate) creates superposition; a $\pi$-rotation (NOT gate) flips qubits. Understanding time evolution is essential for designing quantum circuits.

**In atomic and molecular physics:** Electronic transitions, spectroscopy, and chemical reactions all involve time-dependent evolution of quantum systems. Magnetic resonance (NMR, ESR) uses resonant driving fields to manipulate nuclear and electron spins.

**In fundamental physics:** The Schrödinger equation is arguably the most important equation in physics. All phenomena, from atoms to stars, ultimately obey this fundamental law (though in realistic systems, we must account for interactions with the environment, leading to decoherence).



### Key Concepts

- **Schrödinger equation**: $\mathrm{i}\hbar \frac{\partial|\psi\rangle}{\partial t} = \hat{H}|\psi\rangle$, connecting time evolution to the Hamiltonian (energy operator).

- **Time-evolution operator**: $\hat{U}(t) = \mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$ is a **unitary operator** that directly solves the Schrödinger equation. Unitarity ensures normalization and probability conservation.

- **Stationary states**: Energy eigenstates $|E_n\rangle$ are stationary—they evolve only by a global phase $\mathrm{e}^{-\mathrm{i}E_n t/\hbar}$, which is unobservable. Physical quantities remain constant.

- **Superpositions and interference**: Superpositions of energy eigenstates exhibit quantum interference due to different phase evolution rates. This produces oscillatory behavior, including Rabi oscillations.

- **Larmor precession**: A spin in a magnetic field precesses around the field direction at the **Larmor frequency** $\omega_0 = \gamma B_0$. The Bloch vector rotates around the field axis.

- **Rabi oscillations**: When a two-level system is driven by an oscillating field (resonant with the transition), the population oscillates periodically between the two levels. The **Rabi frequency** $\Omega$ determines the oscillation rate.

- **Heisenberg picture**: Alternative formulation where operators evolve and states are static. The Heisenberg equation $\frac{\mathrm{d}\hat{O}}{dt} = \frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{O}]$ governs operator dynamics.

- **Ehrenfest theorem**: Expectation values follow classical equations of motion. This is the link between quantum and classical mechanics.


### Learning Objectives

By the end of this section, you should be able to:

- State and solve the time-dependent Schrödinger equation for simple Hamiltonians
- Understand why the time-evolution operator must be unitary and what that ensures
- Recognize stationary states and explain why they don't evolve physically
- Compute the time evolution of superpositions of energy eigenstates
- Explain Larmor precession: a spin's Bloch vector rotates around the magnetic field direction
- Understand Rabi oscillations and compute transition probabilities
- Apply the Rabi formula to design quantum gates (e.g., $\pi/2$-pulse, $\pi$-pulse)
- Distinguish between the Schrödinger and Heisenberg pictures and use both
- Apply the Heisenberg equation to find conserved quantities and operator time-dependence
- State and use the Ehrenfest theorem to connect quantum to classical dynamics

## Projects

### Project: Pulse Shaping and GRAPE Optimization for Quantum Gate Design

**Objective:** Develop optimal control techniques (GRAPE: Gradient Ascent Pulse Engineering) to design smooth, robust pulse sequences that implement target quantum gates under realistic Hamiltonian constraints and decoherence.

**Background:** Ideal quantum gates assume infinite control bandwidth and zero error. In practice, control pulses must be smooth, band-limited, and robust to parameter variations. GRAPE and similar optimal control methods are at the frontier of quantum device engineering. They solve the inverse problem: given a target unitary, find the control fields $\Omega(t)$ that generate it under the true microscopic Hamiltonian, accounting for decoherence and driving constraints. This project deepens understanding of time evolution (1.3) by solving it in the presence of realistic imperfections.

**Suggested Approach:**
- **Literature survey**: Study optimal control in quantum systems. Read "GRAPE: A Practical Approach to Quantum Optimal Control" or similar. Understand how gradient descent optimizes fidelity with respect to control parameters.
- **Hamiltonian model**: Work with a spin in a magnetic field: $\hat{H} = \frac{\omega_0}{2}\hat{\sigma}^z + \Omega(t)\hat{\sigma}^x$ (resonant drive, rotating frame). Include a decoherence term: dephasing rate $\gamma_\phi$, spontaneous emission $\gamma_1$.
- **Optimization problem**: Minimize the infidelity $\mathcal{I} = 1 - \frac{1}{2}|\mathrm{Tr}(U_{\text{target}}^\dagger U_{\text{achieved}})|$ by choosing the pulse shape $\Omega(t)$.
- **Implementation**: Code a GRAPE solver using gradient-based optimization (scipy.optimize). Parametrize the pulse as a sum of basis functions (e.g., Fourier components or piecewise constant intervals). Solve the Liouville-von Neumann equation numerically with decoherence.
- **Benchmarks**: Design pulses for: (a) $\pi/2$ rotation (Hadamard-like), (b) controlled NOT gate (using 2-qubit coupling), (c) resilience to pulse amplitude errors.
- **Analysis**: Compare GRAPE-optimized pulses vs. simple analytical pulses. Show how decoherence time $T_2$ sets the shortest achievable gate time. Discuss scalability.

**Expected Deliverable:** Research report (6-8 pages) with: optimal control theory overview, GRAPE algorithm derivation, numerical results for 1-2 qubit gates, fidelity vs. gate time trade-offs, robustness to parameter variations, and discussion of experimental implementation on current quantum devices.

**Key References:** Khaneja et al. (2005) "Optimal control of coupled spin dynamics"; "Optimal Control Methods for Quantum Systems"; recent GRAPE implementations on superconducting qubits (IBM, Google).
