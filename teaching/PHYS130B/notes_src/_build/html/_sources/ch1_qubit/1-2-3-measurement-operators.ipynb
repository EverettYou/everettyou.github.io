## Lecture Notes

The measurement postulate (§1.2.1) tells us that measuring an observable $\hat{O}$ yields one of its eigenvalues $m$, with probability given by the Born rule, and the state collapses to the corresponding eigenstate. Here we reformulate all of this in terms of **measurement operators** (projectors), which provide a cleaner and more powerful language—one that generalizes naturally to degenerate spectra and mixed states.

### Measurement Operators as Projectors

Given an observable $\hat{O}$ with spectral decomposition

$$


\hat{O} = \sum_m m\, |O{=}m\rangle\langle O{=}m|

$$


we define the **measurement operator** for outcome $m$ as the projector onto the corresponding eigenspace:

$$


P_{O=m} = |O{=}m\rangle\langle O{=}m|

$$ (eq-measurement-projector)

:::{admonition} Projector Properties
:class: important

These projectors satisfy three fundamental properties:

- **Hermiticity**: $P_{O=m}^\dagger = P_{O=m}$ — projectors are observables with eigenvalues 0 or 1
- **Idempotence**: $P_{O=m}^2 = P_{O=m}$ — projecting twice is identical to projecting once
- **Completeness**: $\sum_m P_{O=m} = I$ — the projectors for all outcomes partition the Hilbert space

These properties ensure that measurement outcomes are mutually exclusive and exhaustive, and that projectors can represent the measurement process as state collapse.

:::

These projectors satisfy three key properties:

- **Hermiticity**: $P_{O=m}^\dagger = P_{O=m}$
- **Idempotence**: $P_{O=m}^2 = P_{O=m}$ (projecting twice is the same as projecting once)
- **Completeness**: $\sum_m P_{O=m} = I$ (the projectors for all outcomes sum to the identity)

:::{admonition} Spectral Decomposition
:class: important

The observable itself is reconstructed from its projectors via the **spectral decomposition**:

$$


\hat{O} = \sum_m m\, P_{O=m}

$$ (eq-spectral-decomposition-measurement-operators)

This decomposition shows that every Hermitian operator can be expressed as a weighted sum of projectors. The eigenvalues are the weights, and the projectors are the "building blocks" of measurement.

:::

This formula shows that an observable is completely characterized by its eigenvalues $m$ and the corresponding projectors $P_{O=m}$.

:::{admonition} Discussion
:class: dropdown warning

**Can a measurement leave the state unchanged?** Suppose we measure observable $\hat{O}$ on state $|\psi\rangle$ and happen to get outcome $m$ such that $|\psi\rangle$ is already an eigenstate of $\hat{O}$. The projector $P_{O=m}|\psi\rangle = |\psi\rangle$, so state collapse leaves it untouched. Does this mean the measurement caused no disturbance? What if we measure a *different* observable $\hat{O}'$ on the same state and also get an eigenvalue—is measurement "non-disturbing" whenever we happen to measure an eigenbasis? Or is there still a hidden disturbance encoded in the collapse structure itself?
:::

### Explicit Construction: $P_{O=m} = \frac{1 + m\hat{O}}{2}$

For a qubit observable with eigenvalues $m = \pm 1$ (such as any Pauli operator), there is an elegant closed-form expression. Starting from the spectral decomposition $\hat{O} = (+1)|O{=}+\rangle\langle O{=}+| + (-1)|O{=}-\rangle\langle O{=}-|$ and the completeness relation $|O{=}+\rangle\langle O{=}+| + |O{=}-\rangle\langle O{=}-| = I$, we can solve for each projector:

$$


P_{O=m} = \frac{I + m\hat{O}}{2}

$$ (eq-projector-formula)

:::{admonition} Derivation
:class: dropdown note

Adding the two equations:

$$


\hat{O} + I = 2|O{=}+\rangle\langle O{=}+| \implies P_{O=+} = \frac{I + \hat{O}}{2}

$$


Subtracting:

$$


I - \hat{O} = 2|O{=}-\rangle\langle O{=}-| \implies P_{O=-} = \frac{I - \hat{O}}{2}

$$


Both cases are unified by $P_{O=m} = (I + m\hat{O})/2$.
:::

:::{admonition} Verification of Projector
:class: dropdown note

**Idempotence**: Using $\hat{O}^2 = I$ (which holds for any operator with eigenvalues $\pm 1$):

$$


P_{O=m}^2 = \frac{(I + m\hat{O})^2}{4} = \frac{I + 2m\hat{O} + m^2 \hat{O}^2}{4} = \frac{2I + 2m\hat{O}}{4} = P_{O=m} \quad\checkmark

$$


**Completeness**:

$$


P_{O=+} + P_{O=-} = \frac{I + \hat{O}}{2} + \frac{I - \hat{O}}{2} = I \quad\checkmark

$$


**Orthogonality**:

$$


P_{O=+}\, P_{O=-} = \frac{(I+\hat{O})(I-\hat{O})}{4} = \frac{I - \hat{O}^2}{4} = 0 \quad\checkmark

$$


:::

### Measurement Operators for the Pauli Observables

Applying $P_{O=m} = (I + m\hat{O})/2$ to the three Pauli matrices:

**Z-measurement** ($\hat{O} = \hat{\sigma}^z$):

$$


P_{Z=+1} = \frac{I + \hat{\sigma}^z}{2} = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix} = |0\rangle\langle 0|, \qquad P_{Z=-1} = \frac{I - \hat{\sigma}^z}{2} = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix} = |1\rangle\langle 1|

$$


**X-measurement** ($\hat{O} = \hat{\sigma}^x$):

$$


P_{X=+1} = \frac{I + \hat{\sigma}^x}{2} = \frac{1}{2}\begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix} = |{+}\rangle\langle{+}|, \qquad P_{X=-1} = \frac{I - \hat{\sigma}^x}{2} = \frac{1}{2}\begin{pmatrix} 1 & -1 \\ -1 & 1 \end{pmatrix} = |{-}\rangle\langle{-}|

$$


**Y-measurement** ($\hat{O} = \hat{\sigma}^y$):

$$


P_{Y=+1} = \frac{I + \hat{\sigma}^y}{2} = \frac{1}{2}\begin{pmatrix} 1 & -\mathrm{i} \\ i & 1 \end{pmatrix} = |{+}i\rangle\langle{+}i|, \qquad P_{Y=-1} = \frac{I - \hat{\sigma}^y}{2} = \frac{1}{2}\begin{pmatrix} 1 & \mathrm{i} \\ -i & 1 \end{pmatrix} = |{-}i\rangle\langle{-}i|

$$


where $|\pm i\rangle = \frac{1}{\sqrt{2}}(|0\rangle \pm i|1\rangle)$ are the eigenstates of $\hat{\sigma}^y$.

### Measurement Probability as Conditional Probability

The Born rule can now be stated entirely in terms of projectors. The probability of obtaining outcome $m$ when measuring observable $O$ on state $|\psi\rangle$ is:

$$


p(m \mid O, \psi) = \langle\psi| P_{O=m} |\psi\rangle = \|P_{O=m}|\psi\rangle\|^2

$$ (eq-born-rule-projector)

The notation $p(m \mid O, \psi)$ emphasizes that measurement probability is a **conditional probability**: it depends on both the choice of observable $O$ and the state $|\psi\rangle$. Using the projector formula for qubits:

$$


p(m \mid O, \psi) = \left\langle\psi\left|\frac{I + m\hat{O}}{2}\right|\psi\right\rangle = \frac{1 + m\langle\hat{O}\rangle_\psi}{2}

$$ (eq-born-rule-expectation)

where $\langle\hat{O}\rangle_\psi = \langle\psi|\hat{O}|\psi\rangle$ is the expectation value. This gives a beautifully simple relationship: the measurement probability is determined entirely by the expectation value of the observable.

:::{admonition} Discussion
:class: dropdown warning

**Is the projector formalism the end of the story?** We have defined measurement operators as projectors $P_{O=m}$. But is this the most general description of measurement? What if a physical measurement device cannot perfectly distinguish outcomes—say, it detects outcome $m$ with efficiency less than 100%, or with some systematic leakage to other channels? In classical measurement, we would use likelihood functions. In quantum mechanics, this leads to **positive operator-valued measures** (POVMs), where the measurement operators need not be projectors, only positive operators that sum to $I$. Does this mean projectors are only an idealization, and real measurements require a richer framework?
:::

:::{admonition} Example: Measuring $\hat{\sigma}^x$
:class: dropdown note

The expectation value is $\langle 0|\hat{\sigma}^x|0\rangle = 0$, so:

$$


p(+1 \mid X, |0\rangle) = \frac{1 + 0}{2} = \frac{1}{2}, \qquad p(-1 \mid X, |0\rangle) = \frac{1 - 0}{2} = \frac{1}{2}

$$


As expected: a $z$-eigenstate is an equal superposition in the $x$-basis.
:::

### State Collapse as Bayesian Updating

After obtaining outcome $m$, the state collapses according to:

$$


|\psi\rangle \;\xrightarrow{\text{outcome } m}\; |\psi'\rangle = \frac{P_{O=m}|\psi\rangle}{\sqrt{p(m \mid O, \psi)}}

$$ (eq-state-collapse-projector)

The numerator $P_{O=m}|\psi\rangle$ projects the state onto the eigenspace of outcome $m$, keeping only the component compatible with the observed result. The denominator $\sqrt{p(m|O,\psi)}$ renormalizes the state. This is directly analogous to **Bayes' theorem** in classical probability.

In classical probability, Bayes' theorem updates a prior distribution $p(x)$ given observed data $d$:

$$


p(x \mid d) = \frac{p(d \mid x)\, p(x)}{p(d)}

$$


The quantum state collapse follows the same logical structure:

| Classical Bayesian Updating | Quantum State Collapse |
|:---|:---|
| Prior distribution $p(x)$ | Prior state $\|\psi\rangle$ |
| Likelihood $p(d \mid x)$ | Projector $P_{O=m}$ |
| Evidence $p(d) = \sum_x p(d\mid x)p(x)$ | Born probability $p(m\mid O,\psi)$ |
| Posterior $p(x\mid d) \propto p(d\mid x)\,p(x)$ | Post-measurement state $\|\psi'\rangle \propto P_{O=m}\|\psi\rangle$ |

In both cases, the update rule retains only the part of the prior that is **compatible with the observation**, then renormalizes. The key difference is that classical Bayesian updating changes a probability distribution (our knowledge), while quantum state collapse changes the state itself (the physical reality, not just our knowledge of it—at least in the standard interpretation).

:::{admonition} Discussion
:class: dropdown warning

**Where is the quantum-classical boundary in the measurement chain?** The state collapse formula $|\psi'\rangle = P_{O=m}|\psi\rangle / \sqrt{p(m|O,\psi)}$ treats the measurement as an instantaneous, deterministic update. But in the lab, measuring a qubit means interacting it with a physical apparatus (photons, electron beams, etc.). That apparatus is itself quantum. At what point does a quantum interaction *become* a measurement that collapses the state? Is collapse a law of physics, or an emergent property that arises only when we trace out the apparatus degrees of freedom and compute reduced density matrices? Does the answer depend on whether the apparatus is "large enough" to decohere the outcomes?
:::

:::{admonition} Bayesian Measurement Structure
:class: dropdown note

Consider two sequential measurements of observables $O_1$ then $O_2$, obtaining outcomes $m_1$ then $m_2$. The joint probability factorizes as:

$$


p(m_2, m_1 \mid O_2, O_1, \psi) = p(m_2 \mid O_2, \psi'_{m_1})\; p(m_1 \mid O_1, \psi)

$$


where $|\psi'_{m_1}\rangle = P_{O_1=m_1}|\psi\rangle / \sqrt{p(m_1|O_1,\psi)}$ is the post-first-measurement state. This is exactly the **chain rule of conditional probability**, reinforcing the Bayesian interpretation. Each measurement updates the state, and subsequent probabilities are computed from the updated state.
:::

### Degenerate Measurements

When an eigenvalue $m$ is **degenerate** (i.e., multiple linearly independent eigenstates share the same eigenvalue), the projector $P_{O=m}$ projects onto the entire eigenspace, not just a single eigenstate. If the eigenspace for outcome $m$ is spanned by $\{|m, \alpha\rangle\}_{\alpha=1}^{g_m}$ where $g_m$ is the degeneracy, then:

$$


P_{O=m} = \sum_{\alpha=1}^{g_m} |m, \alpha\rangle\langle m, \alpha|

$$ (eq-degenerate-projector)

The measurement probability and state collapse formulas {eq}`eq-born-rule-projector` and {eq}`eq-state-collapse-projector` remain exactly the same—this is the power of the projector formalism. The only difference is that the post-measurement state now lies in a **subspace** rather than being a unique eigenstate:

$$


|\psi'\rangle = \frac{P_{O=m}|\psi\rangle}{\sqrt{p(m|O,\psi)}}

$$


:::{admonition} Why Degeneracy Matters
:class: note

After a degenerate measurement, the state retains coherence **within** the eigenspace. Only the components **outside** the eigenspace are removed. This is sometimes called the **Lüders rule** (as opposed to the von Neumann rule, which would further collapse to a specific eigenstate within the degenerate subspace). The Lüders rule is the minimal disturbance consistent with the measurement outcome—it removes exactly the information gained and nothing more.
:::

:::{admonition} Example: Measuring $L^2$ vs.
:class: dropdown note

Consider the hydrogen atom with orbital angular momentum. Measuring $\hat{L}^2$ with outcome $\ell(\ell+1)\hbar^2$ projects onto the $(2\ell+1)$-dimensional subspace spanned by $\{|l,m\rangle : m = -\ell, \ldots, +\ell\}$. The state after measurement is a superposition within this subspace:

$$


|\psi'\rangle = \frac{\sum_{m=-\ell}^{+\ell} |l,m\rangle\langle l,m|\psi\rangle}{\sqrt{\sum_{m=-\ell}^{+\ell} |\langle l,m|\psi\rangle|^2}}

$$


A subsequent measurement of $\hat{L}_z$ would further collapse this to a specific $|l,m\rangle$. The two measurements together—first $\hat{L}^2$, then $\hat{L}_z$—are compatible (since $[\hat{L}^2, \hat{L}_z]=0$) and together specify the state within the angular momentum sector.
:::

### Summary

- A **measurement operator** $P_{O=m} = |O{=}m\rangle\langle O{=}m|$ is the projector onto the eigenspace of observable $\hat{O}$ with eigenvalue $m$.
- For qubit observables with $m = \pm 1$: $P_{O=m} = (I + m\hat{O})/2$.
- The **Born rule** in projector form: $p(m|O,\psi) = \langle\psi|P_{O=m}|\psi\rangle$.
- **State collapse** is projection followed by renormalization: $|\psi'\rangle = P_{O=m}|\psi\rangle / \sqrt{p}$, analogous to Bayesian updating.
- For **degenerate** eigenvalues, $P_{O=m}$ projects onto the full eigenspace, and the post-measurement state retains coherence within that subspace (Lüders rule).
