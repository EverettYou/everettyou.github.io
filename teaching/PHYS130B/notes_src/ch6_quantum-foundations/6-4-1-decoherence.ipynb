## Lecture Notes

Why do we never see macroscopic superpositions—a cat that is both alive and dead? When a quantum system couples to its environment, entanglement with countless environmental degrees of freedom causes the delicate quantum coherences to decay away, leaving only a classical mixture of states. Decoherence is the bridge between quantum and classical: the process by which the quantum world becomes invisible to us.

### Decoherence Process

Consider a system S (e.g., qubit) entangled with environment E (bath of many modes). The total state is initially:

$$


|\Psi\rangle_{SE} = (\alpha|{0}\rangle_S + \beta|{1}\rangle_S) \otimes |E_0\rangle_E

$$ (eq-decoherence-initial-state)

After interaction for time $t$, the system entangles with different environmental states:

$$


|\Psi(t)\rangle_{SE} = \alpha|{0}\rangle_S |E_0(t)\rangle_E + \beta|{1}\rangle_S |E_1(t)\rangle_E

$$ (eq-decoherence-entangled-state)

where $|E_0(t)\rangle$ and $|E_1(t)\rangle$ are orthogonal (or nearly orthogonal) environmental states.

**Key**: The environment has become **entangled** with the system, encoding which state the system is in ("which-path" information).


:::{admonition} Pointer States and Einselection
:class: important

**Einselection** (environment-induced superselection) is the mechanism by which an environment selects certain states—called **pointer states**—that are most robust to decoherence.

**Pointer states** are eigenstates of the system-environment interaction Hamiltonian that decohere most slowly (or not at all). They correspond to the classical states we observe in nature.

**Mathematical criterion**: A state $|\psi_j\rangle$ is a pointer state if the system-environment coupling $H_\text{int}$ preserves it:

$$

\langle \psi_j | H_\text{int} | \psi_j \rangle \text{ is large relative to off-diagonal coupling}

$$

**Physical consequence**: A superposition of pointer states decoheres rapidly into a classical mixture, but each pointer state remains pure (under unitary evolution alone). This explains why we observe classical outcomes despite quantum mechanics being fundamental.

:::


### Reduced Density Matrix and Decoherence

The **reduced density matrix** of the system (after tracing out environment) is:

$$


\rho_S(t) = \text{Tr}_E(|\Psi(t)\rangle\langle\Psi(t)|) = \rho_\text{diag} + \rho_\text{coherence}

$$ (eq-decoherence-reduced)

Computing the partial trace:

$$


\rho_S(t) = |\alpha|^2 |{0}\rangle\langle{0}| + |\beta|^2 |{1}\rangle\langle{1}| + \alpha^*\beta \mathrm{e}^{-t/T_2} |{0}\rangle\langle{1}| + \beta^*\alpha \mathrm{e}^{-t/T_2} |{1}\rangle\langle{0}|

$$ (eq-decoherence-reduced-explicit)

where the **coherence factor** is:

$$


\mathrm{e}^{-t/T_2} = \langle E_0(t) | E_1(t) \rangle

$$ (eq-decoherence-coherence-factor)

**Interpretation**: The overlap $\langle E_0(t) | E_1(t) \rangle$ measures how distinguishable the environmental states are. As they diverge (become orthogonal), the coherence decays exponentially.

### Exponential Decay: Timescale

In the **weak coupling limit** (system weakly interacts with large environment):

$$


\langle E_0(t) | E_1(t) \rangle \approx \mathrm{e}^{-\gamma t}

$$ (eq-decoherence-exponential)

where $\gamma$ is the decoherence rate. The **decoherence timescale** (dephasing time) is:

$$


T_2 = 1/\gamma

$$ (eq-decoherence-time-scale)

Off-diagonal elements decay as:

$$


\rho_{01}(t) = \alpha^*\beta \mathrm{e}^{-t/T_2}

$$ (eq-decoherence-coherence-decay)

:::{admonition} Pointer States and Einselection
:class: important

**Pointer states**: Special quantum states that are **robust** against decoherence; they remain pure (or nearly pure) even when entangled with the environment.

**Zurek's concept**: The environment **selects** a preferred basis (pointer basis) by rapidly decohering superpositions in that basis.

**Definition**: States that diagonalize the system-environment coupling:
:::

**Definition**: States that diagonalize the system-environment coupling:

$$


A_S = \sum_i a_i |i\rangle\langle i| \quad \text{(diagonal)}

$$


The pointer basis is the **eigenbasis of $A_S$**.

#### Why Pointer States Are Robust

If the system is in pointer state $|i\rangle$, the system-environment interaction gives:

$$


H_\text{int} |i\rangle \otimes |E_k\rangle = a_i |i\rangle \otimes B_{E,k} |E_k\rangle

$$


The environmental state shifts but **doesn't entangle** with different system states (system state remains definite). Only superposition of pointer states decoheres.

#### Einselection (Environment-Induced Selection)

**Zurek's concept**: The environment **selects** a preferred basis (pointer basis) by rapidly decohering superpositions in that basis.

**Mechanism**:
1. System in arbitrary superposition $|\psi\rangle = \sum_i c_i |i\rangle$ (not necessarily pointer basis).
2. Interaction with environment entangles system with different environmental states for each component.
3. Environment becomes a **which-path detector**: stores information about which component the system is in.
4. Superposition of pointer states collapses effectively to a **mixture**.
5. **Result**: Classical-looking mixture in pointer basis.

### Quantum-to-Classical Transition

**Question**: How does a quantum superposition become a classical mixture?

**Answer**: Via decoherence and einselection.

#### Single Pointer State

If system is in pointer state (e.g., $|{0}\rangle$ or $|{1}\rangle$), it evolves **deterministically** under environment interaction (no randomness):

$$


\rho_S(t) = |{0}\rangle\langle{0}| \quad \text{(pure state, no decoherence)}

$$


#### Superposition of Pointer States

If system is in superposition $|\psi\rangle = \frac{1}{\sqrt{2}}(|{0}\rangle + |{1}\rangle)$:

$$


\rho_S(t) \approx \begin{pmatrix} 1/2 & \mathrm{e}^{-t/T_2}/2 \\ \mathrm{e}^{-t/T_2}/2 & 1/2 \end{pmatrix} \to \frac{I}{2}

$$


As $t \gg T_2$, the coherence vanishes; the state **looks like** a classical 50/50 mixture:

$$


\rho_S(\infty) = \frac{1}{2}|{0}\rangle\langle{0}| + \frac{1}{2}|{1}\rangle\langle{1}|

$$


**But**: The full system+environment state is still **pure** (reversible). The apparent classicality is only from the system's perspective; the environment has the information.

#### Macroscopic Objects

For macroscopic superpositions (Schrödinger cat): $|\psi\rangle = \frac{1}{\sqrt{2}}(|\text{alive}\rangle + |\text{dead}\rangle)$, the decoherence rate is **enormous**:

$$


\gamma \sim 10^{23} \text{ s}^{-1} \Rightarrow T_2 \sim 10^{-23}\text{ s}

$$


Macroscopic superpositions decohere **almost instantaneously**; this is why we never observe them.

### Decoherence-Free Subspaces (DFS)

**Exception**: Some quantum states can be **protected** from certain decoherence channels via symmetry.

**Example**: Singlet state $|\Psi^-\rangle = \frac{1}{\sqrt{2}}(|{01}\rangle - |{10}\rangle)$.

If both qubits undergo **identical** amplitude damping (same coupling $g_k$), the singlet state is **immune** (decoherence-free):

$$


\hat{\sigma}_-^{(1)} + \hat{\sigma}_-^{(2)} \text{ acting on } |\Psi^-\rangle = 0

$$


**Application**: Quantum error correction; storing information in entangled subspaces.

### Dynamical Decoupling

**Idea**: Apply **rapidly oscillating pulses** to the system to average out environmental coupling and suppress decoherence.

**Example (CPMG sequence)**: Apply $\pi$-pulses at regular intervals. If the control field is much faster than environment:  the effective interaction averages to zero.

**Result**: Coherence time extended: $T_2^\text{eff} >> T_2^\text{bare}$.

**Trade-off**: Requires rapid, accurate pulses (control overhead).

### Timescale Hierarchy

In typical systems:

$$


T_\text{system} >> T_2 >> T_1

$$


where:
- $T_\text{system}$ ~ inverse Rabi frequency (system evolution timescale).
- $T_2$ ~ dephasing time (coherence decay); environment-induced randomness.
- $T_1$ ~ relaxation time (energy dissipation); thermal equilibration.

**Practical consequence**: Decoherence (via dephasing) is often the **dominant** limitation before energy relaxation becomes significant.

:::{admonition} Decoherence is not collapse
:class: caution
Decoherence explains why macroscopic superpositions are never observed—off-diagonal elements of $\rho$ decay on timescale $T_2$. But decoherence alone does not select a single outcome; it only suppresses interference between branches. The measurement problem remains.
:::

### Quantum Darwinism: Emergence of Classicality via Environmental Redundancy

**Concept** (Zurek): Beyond decoherence lies Quantum Darwinism—the proliferation of information about a quantum system throughout the environment. Multiple environmental degrees of freedom independently acquire information about the system's state, enabling multiple observers to determine which state the system is in **without directly accessing it**. This redundant encoding explains why we see a **classical world** emerging from quantum mechanics.

#### Einselection Revisited: Environment-Induced Superselection

Recall that einselection picks out the **pointer states** as special: the environment selects a preferred basis because superpositions in that basis rapidly decohere. More fundamentally, einselection is a **superselection rule** imposed by the environment.

**Definition**: A superselection rule forbids certain coherences (off-diagonal elements of the density matrix) from being created or detected by the environment.

**Mechanism**:
1. System initially in arbitrary superposition $|\psi\rangle = \sum_i c_i |i\rangle$ (arbitrary basis).
2. System couples to environment; each component $|i\rangle$ entangles with **distinct** environmental state $|E_i\rangle$.
3. The system-environment state becomes:

$$


| \Psi \rangle_{SE} = \sum_i c_i |i\rangle \otimes |E_i\rangle

$$


4. Environmental states are **orthogonal** (or nearly so): $\langle E_i | E_j \rangle \approx \delta_{ij}$.
5. The environment has become a **which-path detector**: by measuring the environment, any external agent can determine which component the system is in.
6. Coherences between different pointer states are **unobservable** from the system's perspective (traced-out environment contains the "which-path" information).
7. **Result**: The pointer basis is the only basis in which the system appears as a classical mixture; superposition in that basis effectively disappears.

#### Redundant Information: Multiple Copies in the Environment

**Key insight**: The environment is **large**—many degrees of freedom. Information about which pointer state the system is in doesn't get stored in just one environmental mode; it spreads across countless modes.

**Redundancy mechanism**:
1. System in pointer state $|i\rangle$ interacts with environment.
2. A small environmental subsystem $E_1$ (single photon, atom, phonon mode) becomes entangled: can carry "fingerprint" of state $|i\rangle$.
3. Another subsystem $E_2$ subsequently interacts; it also becomes entangled and acquires the same information.
4. This repeats: $E_3, E_4, \ldots$ all gradually learn which state the system is in.
5. **Result**: $N$ independent environmental fragments, each carrying (partial) information about the system state.

**Classical signature**: In classical physics, we don't worry about copies. Many independent observers can measure the same property (e.g., the temperature of a room) simultaneously; the environment stores this information redundantly. Quantum Darwinism shows that **quantum information also becomes redundant** for pointer states, mimicking classicality.

#### Quantifying Redundancy: Quantum Mutual Information

**Mutual information** $I(S : E_k)$ between system $S$ and environmental fragment $E_k$ quantifies how much information the fragment carries about the system:

$$


I(S : E_k) = S(\rho_S) + S(\rho_{E_k}) - S(\rho_{S E_k})

$$


where $S(\rho) = -\text{Tr}(\rho \log_2 \rho)$ is the von Neumann entropy.

**Interpretation**:
- $I(S : E_k) = 0$: Fragment $E_k$ carries no information about $S$.
- $I(S : E_k) > 0$: Fragment $E_k$ carries some information about $S$.
- $I(S : E_k) \approx S(\rho_S)$: Fragment $E_k$ nearly determines the system state (maximum redundancy).

**Darwinian redundancy**: Total information in $N$ fragments:

$$


\mathcal{R} = \sum_{k=1}^N I(S : E_k)

$$


**Classical case** (e.g., room temperature): If $N$ observers independently measure the temperature, each acquires the full information (entropy $\sim \log 2$ states). Total $\mathcal{R} \sim N$—massive redundancy.

**Quantum pointer states**: As the system and environment interact, $\mathcal{R}$ grows. Multiple environmental fragments each learn the pointer state; **redundancy increases exponentially with environment size**. This is the hallmark of Darwinism.

#### Objectivity: Why the Classical World Emerges

**Definition**: A property is **objective** if multiple independent observers measure it and obtain the same result.

**Classical objectivity**: A thermometer reads 20 °C. Another observer, measuring at the same location, also reads 20 °C. Property is objective because it's independent of who measures.

**Quantum Darwinism explanation**:
1. System in pointer state $|i\rangle$ (e.g., spin up).
2. Observer Alice measures a small environmental fragment; she learns the system is in state $|i\rangle$.
3. Observer Bob, independently measuring a **different** environmental fragment, also learns the system is in state $|i\rangle$.
4. They agree on the result; the property (system state) is **objective**.
5. This works because both fragments carry the same information (redundancy).

**Non-objectivity in superposition**:
If the system were in superposition $|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$:
- Alice measures fragment $E_1$, learns state 0.
- Bob measures fragment $E_2$, learns state 1.
- They disagree; there is no objective state.
- However, with decoherence and Darwinism, the superposition rapidly evolves into a mixture where state $|0\rangle$ or $|1\rangle$ is definite (though we don't know which without measuring). Redundancy ensures any observer learns the same state.

**Classical world emerges**: The pointer states become **objective observables**. Superpositions are not objective (not accessible to multiple independent observers), so they effectively don't exist in the realm we observe.

#### Connection to Decoherence: Darwinism as the Next Step

**Decoherence (local view)**:
- System's density matrix: coherences decay away ($\rho_{ij} \to 0$ for $i \ne j$).
- Time scale: $T_2 \sim 1/\gamma$ (depends on coupling strength and environment size).
- **Question unanswered**: Why do we observe definite states, not a statistical ensemble of unknown states?

**Quantum Darwinism (global view)**:
- Explains WHY pointer states are special: information about them spreads redundantly.
- Explains OBJECTIVITY: multiple observers independently learn the same pointer state from different environmental fragments.
- Explains CLASSICALITY: redundancy prevents quantum coherences from being recovered; pointer states appear classical.
- Connects decoherence to observation: decoherence of superpositions is equivalent to proliferation of information about pointer states throughout the environment.

**Quantitative relation**:

As decoherence progresses (off-diagonal $\rho_{ij}$ decay), mutual information $I(S : E_k)$ grows:

$$


\frac{\mathrm{d}I(S : E_k)}{dt} \propto \gamma \quad \text{(Darwinian growth rate)}

$$


The faster decoherence (larger $\gamma$), the faster Darwinian redundancy builds up. Decoherence erases quantum coherence **while simultaneously encoding information redundantly** throughout the environment.

#:::{admonition} Example: Two-Level System Decoherence
:class: dropdown note

A two-level system (qubit, spin) is in superposition $|\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$.

**Pointer states**: $|0\rangle$ and $|1\rangle$ (energy eigenstates).

**Environment**: Blackbody radiation at temperature $T$; infinite modes.

**Interaction**:
- Qubit in state $|1\rangle$ emits a photon (decay); transitions to $|0\rangle$.
- Each photon carries information: "qubit was in $|1\rangle$".
- Different photons encode the same information (redundancy).

**Darwinism**:
1. First photon emitted; fragment $E_1$ learns state is $|1\rangle$.
2. Second photon emitted; fragment $E_2$ learns state is $|1\rangle$.
3. $N$-th photon: fragment $E_N$ also learns state is $|1\rangle$.
4. Any observer measuring a photon learns "qubit is in $|1\rangle$".
5. Objectivity: all observers agree.

**Decoherence side**: As more photons are emitted, the qubit's state becomes increasingly mixed (decay). Superposition decoheres. This is the local manifestation.

**Darwinism side**: Simultaneously, each emitted photon carries the same information; massive redundancy. The environment "has learned" which state the qubit is in; classicality emerges.

:::

#### Limits of Darwinism: Pure Einselection

**Not all states are pointer states**: States that don't decohere (decoherence-free subspaces, quantum error correcting codes) don't undergo Darwinism.

**Information in non-pointer basis**: If the system is in a superposition of non-pointer states (e.g., entangled state $|\Psi^-\rangle$), the environment does not redundantly encode information in the usual sense. To learn the state requires global measurements (on full environment); single environmental fragments carry minimal information.

**Quantum advantage**: This is the basis of quantum error correction and quantum information processing—information can be hidden from local environmental access via entanglement.


### Summary

- **Decoherence**: Interaction with environment entangles system with environmental degrees of freedom; reduced density matrix becomes mixed; coherences decay exponentially.
- **Mechanism**: System-environment entanglement encodes which-path information; environment becomes a detector; quantum interference is erased.
- **Timescale**: Decoherence time $T_2 = 1/\gamma$; depends on system size, environment temperature, coupling strength.
- **Pointer states**: Eigenstates of system operator in coupling Hamiltonian; robust against decoherence.
- **Einselection**: Environment selects preferred basis (pointer basis); superposition in that basis rapidly decoheres to mixture.
- **Quantum-to-classical**: Coherent superposition → classical mixture; pointer states classically distinguishable; non-pointer basis requires quantum control to maintain.
- **Macroscopic superpositions**: Decohere almost instantaneously due to many environmental couplings.
- **Decoherence-free subspaces**: Symmetry-protected states immune to certain channels; used in quantum error correction.
- **Dynamical decoupling**: Rapid control pulses suppress decoherence via refocusing; extends coherence time.
