## Lecture Notes

For an isolated system, the Schrödinger equation perfectly captures quantum evolution. But real systems interact with their surroundings — heat baths, photon fields, uncontrolled degrees of freedom. The Lindblad master equation describes how open systems evolve, accounting for both unitary dynamics and the irreversible effects of environmental coupling. It is the most general form of Markovian (memoryless) evolution compatible with quantum mechanics.

### Overview

The Lindblad equation decomposes open-system evolution into two parts: (1) unitary evolution under the system Hamiltonian $\hat{H}$, and (2) dissipation described by jump operators $\{\hat{L}_k\}$ with rates $\{\gamma_k\}$. The equation preserves trace, Hermiticity, and positivity — necessary conditions for valid quantum dynamics. Every quantum channel (§6.3.3) is obtained from a Lindblad equation integrated over a fixed time interval.

:::{admonition} Discussion: Markovian Approximation
:class: dropdown tip

The Lindblad equation assumes the system and environment are uncorrelated at each instant — a Markovian (memoryless) approximation. But real environments can have memory: the environment's response depends on the system's past state. How do we know when Markovian is justified?

- The approximation breaks down if the environment correlation time $\tau_E$ is comparable to the system timescale. For typical quantum optics setups, $\tau_E \sim 10^{-15}$ s while atomic evolution is $\sim 10^{-9}$ s — safely Markovian.
- Some systems show "non-Markovian revivals" — coherence temporarily returns after decoherence. Does this invalidate the Lindblad equation, or is it simply outside its regime?
:::

### Lindblad Master Equation

For an open system coupled to an environment, the reduced density matrix $\hat{\rho}_S$ evolves as:

:::{admonition} Lindblad Master Equation (GKSL Form)
:class: important

$$

\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{\rho}] + \sum_k \gamma_k \left( \hat{L}_k \hat{\rho} \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k, \hat{\rho}\} \right)

$$ (eq-lindblad-equation)

where:
- $H$ = system Hamiltonian (Hermitian), governing unitary evolution.
- $\hat{L}_k$ = **Lindblad operators** (jump operators); generally non-Hermitian.
- $\gamma_k \geq 0$ = **decay rates** (non-negative real numbers).
- $\{A, B\} = AB + BA$ = anticommutator.

This is the most general Markovian master equation compatible with complete positivity and trace preservation (Lindblad 1976; Kossakowski, Sudarshan — GKSL form).
:::

The Lindblad equation is derived by tracing out the environment from unitary system+environment evolution:

:::{admonition} Derivation: Lindblad Equation from the GKSL Theorem
:class: dropdown information

**Goal:** Derive the most general time-local (Markovian) master equation for the density matrix of an open quantum system.

**Setup: Quantum dynamical semigroup.** Markovian evolution means the map $\Lambda_t : \hat{\rho}(0) \mapsto \hat{\rho}(t)$ satisfies the **semigroup property**:

$$
\Lambda_{t+s} = \Lambda_t \circ \Lambda_s, \qquad \Lambda_0 = \mathrm{id}.
$$

(Past history is irrelevant; only the current state matters.) Each $\Lambda_t$ must be a CPTP (completely positive, trace-preserving) map. The generator $\mathcal{L}$ is defined by

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = \mathcal{L}(\hat{\rho}) = \lim_{\varepsilon\to 0^+}\frac{\Lambda_\varepsilon(\hat{\rho}) - \hat{\rho}}{\varepsilon}.
$$

**Step 1: Kraus expansion for infinitesimal $\varepsilon$.** By the Kraus theorem, every CPTP map $\Lambda_\varepsilon$ has a Kraus decomposition:

$$
\Lambda_\varepsilon(\hat{\rho}) = \sum_\mu K_\mu(\varepsilon)\,\hat{\rho}\,K_\mu^\dagger(\varepsilon), \qquad \sum_\mu K_\mu^\dagger K_\mu = \hat{I}.
$$

For small $\varepsilon$, one Kraus operator is close to the identity while the others vanish:

$$
K_0(\varepsilon) = \hat{I} + \varepsilon \hat{F} + O(\varepsilon^2), \qquad
K_k(\varepsilon) = \sqrt{\varepsilon}\,\hat{L}_k + O(\varepsilon^{3/2}), \quad k \geq 1.
$$

The scaling $K_k \sim \sqrt{\varepsilon}$ (not $\sim \varepsilon$) ensures that the $K_k$-terms contribute at $O(\varepsilon)$ when squared.

**Step 2: Trace-preservation constraint.** The condition $\sum_\mu K_\mu^\dagger K_\mu = \hat{I}$ at order $\varepsilon$:

$$
K_0^\dagger K_0 + \sum_{k\geq 1} K_k^\dagger K_k = \hat{I}.
$$

Substituting:

$$
(\hat{I} + \varepsilon\hat{F}^\dagger)(\hat{I} + \varepsilon\hat{F}) + \varepsilon\sum_k \hat{L}_k^\dagger\hat{L}_k = \hat{I} + O(\varepsilon^2).
$$

At order $\varepsilon$: $\hat{F} + \hat{F}^\dagger + \sum_k \hat{L}_k^\dagger\hat{L}_k = 0$.

Decompose $\hat{F}$ into Hermitian and anti-Hermitian parts: $\hat{F} = -\mathrm{i}\hat{H}/\hbar - \hat{\Gamma}/2$, where $\hat{H} = \hat{H}^\dagger$ (effective system Hamiltonian) and $\hat{\Gamma} = \hat{\Gamma}^\dagger$. Then $\hat{F} + \hat{F}^\dagger = -\hat{\Gamma}$, so the constraint gives:

$$
\hat{\Gamma} = \sum_k \hat{L}_k^\dagger\hat{L}_k \quad\Rightarrow\quad
K_0(\varepsilon) = \hat{I} + \varepsilon\left(-\frac{\mathrm{i}\hat{H}}{\hbar} - \frac{1}{2}\sum_k \hat{L}_k^\dagger\hat{L}_k\right) + O(\varepsilon^2).
$$

**Step 3: Compute the generator.** Assemble $\Lambda_\varepsilon(\hat{\rho}) = K_0\hat{\rho}K_0^\dagger + \varepsilon\sum_k \hat{L}_k\hat{\rho}\hat{L}_k^\dagger + O(\varepsilon^2)$:

$$
K_0\hat{\rho}K_0^\dagger = \hat{\rho} + \varepsilon\!\left(\hat{F}\hat{\rho} + \hat{\rho}\hat{F}^\dagger\right) + O(\varepsilon^2)
= \hat{\rho} + \varepsilon\!\left(-\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] - \frac{1}{2}\sum_k\!\{\hat{L}_k^\dagger\hat{L}_k, \hat{\rho}\}\right) + O(\varepsilon^2).
$$

Adding the jump terms $\varepsilon\sum_k \hat{L}_k\hat{\rho}\hat{L}_k^\dagger$:

$$
\Lambda_\varepsilon(\hat{\rho}) = \hat{\rho} + \varepsilon\left[-\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] + \sum_k\left(\hat{L}_k\hat{\rho}\hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger\hat{L}_k,\hat{\rho}\}\right)\right] + O(\varepsilon^2).
$$

Dividing by $\varepsilon$ and taking $\varepsilon\to 0$ yields the **GKSL (Lindblad) master equation**:

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] + \sum_k\left(\hat{L}_k\hat{\rho}\hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger\hat{L}_k,\hat{\rho}\}\right).
$$

**Physical interpretation of each term:**
- $-\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}]$: unitary evolution under the system Hamiltonian (includes any Lamb-shift from the bath).
- $\hat{L}_k\hat{\rho}\hat{L}_k^\dagger$: **quantum jumps** — inflow of probability from the $k$-th dissipation channel.
- $-\frac{1}{2}\{\hat{L}_k^\dagger\hat{L}_k,\hat{\rho}\}$: **no-jump evolution** — outflow of probability maintaining trace preservation.

**Uniqueness:** The decomposition into $\hat{H}$ and $\{\hat{L}_k\}$ is not unique (unitary freedom in the Lindblad operators and basis changes in jump-operator space), but the physical dynamics is unique.
:::

### Properties of Lindblad Evolution

:::{admonition} Properties of the Lindblad Equation
:class: note

The Lindblad equation preserves all fundamental properties of quantum states:

**1. Trace preservation**: $\frac{\mathrm{d}}{\mathrm{d}t}\text{Tr}(\hat{\rho}) = 0$, so $\text{Tr}(\hat{\rho}(t)) = 1$ for all $t$.

**2. Hermiticity preservation**: If $\hat{\rho}(0) = \hat{\rho}(0)^\dagger$, then $\hat{\rho}(t) = \hat{\rho}(t)^\dagger$.

**3. Positivity**: If $\hat{\rho}(0) \geq 0$, then $\hat{\rho}(t) \geq 0$ for all $t$ (valid quantum state).

**4. CPTP map**: The time evolution $\hat{\rho}(0) \mapsto \hat{\rho}(t)$ is a completely positive trace-preserving (quantum) channel, equivalent to the Kraus representation $\hat{\rho}(t) = \sum_\mu K_\mu \hat{\rho}(0) K_\mu^\dagger$ with $\sum_\mu K_\mu^\dagger K_\mu = I$.
:::


:::{admonition} Derivation: Trace Preservation and Dissipation Rates
:class: dropdown information

**Trace preservation.** Compute $\frac{\mathrm{d}}{\mathrm{d}t}\operatorname{Tr}(\hat{\rho})$ using the Lindblad equation $\dot{\hat{\rho}} = -\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] + \sum_k\gamma_k\mathcal{D}[\hat{L}_k]\hat{\rho}$, where $\mathcal{D}[\hat{L}]\hat{\rho} = \hat{L}\hat{\rho}\hat{L}^\dagger - \frac{1}{2}\{\hat{L}^\dagger\hat{L},\hat{\rho}\}$.

The Hamiltonian term: $\operatorname{Tr}([\hat{H},\hat{\rho}]) = \operatorname{Tr}(\hat{H}\hat{\rho} - \hat{\rho}\hat{H}) = 0$ (cyclicity of trace).

Each jump term:

$$
\operatorname{Tr}(\mathcal{D}[\hat{L}]\hat{\rho}) = \operatorname{Tr}(\hat{L}\hat{\rho}\hat{L}^\dagger) - \frac{1}{2}\operatorname{Tr}(\hat{L}^\dagger\hat{L}\hat{\rho}) - \frac{1}{2}\operatorname{Tr}(\hat{\rho}\hat{L}^\dagger\hat{L}) = \operatorname{Tr}(\hat{L}^\dagger\hat{L}\hat{\rho}) - \operatorname{Tr}(\hat{L}^\dagger\hat{L}\hat{\rho}) = 0
$$

using cyclicity. So $\frac{\mathrm{d}}{\mathrm{d}t}\operatorname{Tr}(\hat{\rho}) = 0$ — trace is exactly preserved.

**Positivity of dissipation rates.** The rates $\gamma_k \geq 0$ are required for positivity of $\hat{\rho}(t)$. Negative rates would cause population inversion without input, violating the second law. The Markovian approximation ensures $\gamma_k > 0$ when the environment is in a thermal state at positive temperature.

**Equilibrium state.** At long times, $\dot{\hat{\rho}} = 0$. For a single mode at rate $\gamma$ with jump operator $\hat{L} = \hat{a}$ (amplitude damping), the steady state is the **vacuum**: $\hat{\rho}_{\text{ss}} = \vert 0\rangle\langle 0\vert$. For the thermal Lindblad (both loss $\hat{L}_- = \hat{a}$ and gain $\hat{L}_+ = \hat{a}^\dagger$ at rates $\gamma(\bar{n}+1)$ and $\gamma\bar{n}$), the steady state is the **thermal state** $\hat{\rho}_{\text{th}} = \mathrm{e}^{-\beta\hbar\omega\hat{a}^\dagger\hat{a}}/Z$, as expected.
:::

**Gain and loss balance**: The $\hat{L}_k\hat{\rho} \hat{L}_k^\dagger$ term represents *gain* (probability inflow from quantum jumps), while $-\frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k,\hat{\rho}\}$ represents *loss* (probability outflow). Their combined action preserves trace:

$$

\text{Tr}\left(\hat{L}_k\hat{\rho} \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k,\hat{\rho}\}\right) = \text{Tr}(\hat{L}_k^\dagger \hat{L}_k\hat{\rho}) - \text{Tr}(\hat{L}_k^\dagger \hat{L}_k\hat{\rho}) = 0

$$

### Common Lindblad Operators

#### 1. Amplitude Damping: $L = \sqrt{\Gamma_1}\,\hat{\sigma}_-$

**Physical process**: Excited state $\vert 1\rangle$ decays to ground state $\vert 0\rangle$ (energy dissipation, spontaneous emission) at rate $\Gamma_1 = 1/T_1$.

$$

\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = \Gamma_1\!\left(\hat{\sigma}_-\hat{\rho}\hat{\sigma}_+ - \frac{1}{2}\{\hat{\sigma}_+\hat{\sigma}_-,\hat{\rho}\}\right)

$$ (eq-lindblad-amp-damping)

**Effect on populations**: $\dot{\hat{\rho}}_{11} = -\Gamma_1\hat{\rho}_{11}$ (exponential decay), $\dot{\hat{\rho}}_{00} = +\Gamma_1\hat{\rho}_{11}$ (gain).

**Effect on coherences**: $\dot{\hat{\rho}}_{01} = -\frac{\Gamma_1}{2}\hat{\rho}_{01}$ (decay at half the rate).

#### 2. Phase Damping: $L = \sqrt{\Gamma_\varphi}\,\hat{\sigma}^z$

**Physical process**: Random phase fluctuations (dephasing) with no energy loss at rate $\Gamma_\varphi = 1/T_\varphi$.

$$

\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = \Gamma_\varphi\!\left(\hat{\sigma}^z\hat{\rho}\hat{\sigma}^z - \hat{\rho}\right)

$$ (eq-lindblad-phase-damping)

**Effect**: Only off-diagonal terms decay: $\hat{\rho}_{01}(t) = \mathrm{e}^{-2\Gamma_\varphi t}\hat{\rho}_{01}(0)$. Diagonal (population) terms unchanged. This is the microscopic realization of the dephasing channel (§6.3.3 and §6.4.1).

#### 3. Depolarizing: $L_\mu = \sqrt{\Gamma_d/3}\,\hat{\sigma}^\mu$ for $\mu = x,y,z$

**Physical process**: Symmetric noise on all three axes; generic random error channel at rate $\Gamma_d$.

**Effect**: Drives any state toward the maximally mixed state $I/2$ on timescale $1/\Gamma_d$.

### Steady States

A **steady state** $\hat{\rho}_\infty$ satisfies $\mathcal{L}(\hat{\rho}_\infty) = 0$, i.e., $\frac{\mathrm{d}\hat{\rho}_\infty}{\mathrm{d}t} = 0$.

:::{admonition} Example: Amplitude Damping
:class: dropdown example

**Problem.** Find the steady state for amplitude damping $L = \sqrt{\Gamma_1}\hat{\sigma}_-$.

**Solution.** From $\dot{\hat{\rho}}_{11} = -\Gamma_1\hat{\rho}_{11} = 0$ we get $\hat{\rho}_{11}^\infty = 0$, hence $\hat{\rho}_{00}^\infty = 1$.

The steady state is $\hat{\rho}_\infty = \vert 0\rangle\langle 0\vert$ — all population has decayed to the ground state.
:::

:::{admonition} Example: Thermal Steady State
:class: dropdown example

**Problem.** For an atom coupled to a thermal bath at temperature $T$, include both decay ($L_1 = \sqrt{\Gamma\bar{n}_\text{th}+\Gamma}\,\hat{\sigma}_-$) and absorption ($L_2 = \sqrt{\Gamma\bar{n}_\text{th}}\,\hat{\sigma}_+$). Find the steady state.

**Solution.** Balancing gain and loss: $\hat{\rho}_{11}^\infty/\hat{\rho}_{00}^\infty = \bar{n}_\text{th}/(\bar{n}_\text{th}+1) = \mathrm{e}^{-\hbar\omega/k_BT}$, which is the **Boltzmann factor**. The steady state is the thermal density matrix $\hat{\rho}_\infty = \mathrm{e}^{-\beta \hat{H}}/Z$, consistent with §6.1.2.
:::

:::{admonition} Derivation: Bloch Equations from the Lindblad Equation
:class: dropdown information

We derive the equations of motion for the **Bloch vector** $\boldsymbol{r} = (r_x, r_y, r_z)$, where $r_\mu = \langle\hat\sigma^\mu\rangle = \text{Tr}(\hat\sigma^\mu\hat\rho)$, from the Lindblad equation with amplitude damping and pure dephasing.

**Lindblad operators.** Consider two jump operators:
- $L_1 = \sqrt{\Gamma_1}\,\hat{\sigma}_-$ (amplitude damping, rate $\Gamma_1 = 1/T_1$)
- $L_2 = \sqrt{\Gamma_\varphi/2}\,\hat{\sigma}^z$ (pure dephasing, rate $\Gamma_\varphi$)

**Equation of motion for $\rho_{11}$ (excited state population).**

Using $\hat{\sigma}_+\hat{\sigma}_- = \vert 1\rangle\langle 1\vert$:

$$
\dot\rho_{11} = \Gamma_1(\hat\sigma_-\hat\rho\hat\sigma_+)_{11} - \Gamma_1(\hat\sigma_+\hat\sigma_-\hat\rho)_{11} = \Gamma_1\cdot 0 - \Gamma_1\rho_{11} = -\Gamma_1\rho_{11}
$$

(Dephasing $L_2$ does not change populations.) Therefore the $z$-component relaxes as:

$$
\dot{r}_z = \dot\rho_{00} - \dot\rho_{11} = \Gamma_1\rho_{11} + \Gamma_1\rho_{11} = 2\Gamma_1\frac{1-r_z}{2} = \Gamma_1(1 - r_z)
$$

which gives $r_z(t) = 1 - (1-r_z(0))\mathrm{e}^{-t/T_1}$, decaying toward the ground state $r_z = +1$.

**Equation of motion for $\rho_{01}$ (coherence).**

From amplitude damping: $\dot\rho_{01}^{(1)} = -\tfrac{\Gamma_1}{2}\rho_{01}$ (half the population decay rate damps coherences).

From pure dephasing: $L_2\hat\rho L_2^\dagger = \tfrac{\Gamma_\varphi}{2}\hat\sigma^z\hat\rho\hat\sigma^z$. The off-diagonal element picks up $(-1)(+1) = -1$ from the $\hat\sigma^z$ eigenvalues, so:

$$
\dot\rho_{01}^{(2)} = \tfrac{\Gamma_\varphi}{2}(-1)(+1)\rho_{01} - \tfrac{\Gamma_\varphi}{2}\rho_{01} = -\Gamma_\varphi\rho_{01}
$$

Combining:

$$
\dot\rho_{01} = -\left(\tfrac{\Gamma_1}{2} + \Gamma_\varphi\right)\rho_{01} \equiv -\frac{1}{T_2}\rho_{01}
$$

where $\frac{1}{T_2} = \frac{\Gamma_1}{2} + \Gamma_\varphi$ is the **total dephasing rate**.

**Bloch equations.** Since $r_x = \rho_{01} + \rho_{10}$ and $r_y = \mathrm{i}(\rho_{01} - \rho_{10})$:

$$
\boxed{\dot{r}_x = -\frac{r_x}{T_2}, \quad \dot{r}_y = -\frac{r_y}{T_2}, \quad \dot{r}_z = -\frac{r_z - 1}{T_1}}
$$

with $1/T_1 = \Gamma_1$ (energy relaxation) and $1/T_2 = \Gamma_1/2 + \Gamma_\varphi$ (total transverse decay).

**Physical insight.** The factor of 2: $T_2 \leq 2T_1$. Energy relaxation always contributes $\Gamma_1/2$ to dephasing (population decay randomizes phase); pure dephasing adds extra $\Gamma_\varphi$. The bound $1/T_2 \geq 1/(2T_1)$ (i.e., $T_2 \leq 2T_1$) is fundamental to all qubit experiments.
:::

### Quantum Jump Picture

The Lindblad equation averages over an ensemble of quantum trajectories. Each individual trajectory evolves stochastically:

1. **Between jumps**: The state evolves under the *non-Hermitian* effective Hamiltonian

$$

H_\text{eff} = H - \frac{\mathrm{i}\hbar}{2}\sum_k\gamma_k \hat{L}_k^\dagger \hat{L}_k

$$ (eq-lindblad-heff)

   (non-Hermitian: the norm decreases, tracking the "no-jump" probability.)

2. **At random times** (Poisson process), a **quantum jump** $\hat{L}_k$ occurs: $\hat{\rho} \to \hat{L}_k\hat{\rho} \hat{L}_k^\dagger/p_k$ with probability $p_k \propto \gamma_k\,\text{Tr}(\hat{L}_k^\dagger \hat{L}_k\hat{\rho})\,\mathrm{d}t$.

3. **Averaging** over all trajectories (all possible jump sequences and times) recovers the Lindblad master equation.

**Physical interpretation**: Each quantum jump corresponds to a physical event — photon emission, phonon absorption, spin flip. Decoherence and dissipation arise from random jumps; the deterministic master equation is the statistical average.


:::{admonition} Derivation: Exponential Decay of Coherences Under the Dephasing Channel
:class: dropdown information

**Setup.** The simplest Lindblad equation describes pure **dephasing** of a qubit with jump operator $\hat{L} = \hat{\sigma}^z$ and rate $\gamma$:

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = \gamma\!\left(\hat{\sigma}^z \hat{\rho} \hat{\sigma}^z - \hat{\rho}\right)
$$

(no Hamiltonian term for simplicity). Write $\hat{\rho} = \begin{pmatrix} \rho_{00} & \rho_{01} \\ \rho_{10} & \rho_{11} \end{pmatrix}$ and compute each matrix element using $\hat{\sigma}^z = \begin{pmatrix}1&0\\0&-1\end{pmatrix}$.

**Step 1: Compute $\hat{\sigma}^z\hat{\rho}\hat{\sigma}^z$.**

$$
\hat{\sigma}^z\hat{\rho}\hat{\sigma}^z = \begin{pmatrix}1&0\\0&-1\end{pmatrix}\begin{pmatrix}\rho_{00}&\rho_{01}\\\rho_{10}&\rho_{11}\end{pmatrix}\begin{pmatrix}1&0\\0&-1\end{pmatrix}
= \begin{pmatrix}\rho_{00}&-\rho_{01}\\-\rho_{10}&\rho_{11}\end{pmatrix}
$$

**Step 2: Form the Lindblad superoperator.**

$$
\hat{\sigma}^z\hat{\rho}\hat{\sigma}^z - \hat{\rho} = \begin{pmatrix}0&-2\rho_{01}\\-2\rho_{10}&0\end{pmatrix}
$$

**Step 3: Read off the equations of motion.**

$$
\dot{\rho}_{00} = 0, \quad \dot{\rho}_{11} = 0, \quad \dot{\rho}_{01} = -2\gamma\rho_{01}, \quad \dot{\rho}_{10} = -2\gamma\rho_{10}
$$

**Step 4: Solve.**

- **Diagonal elements** (populations) are **constant**: $\rho_{00}(t) = \rho_{00}(0)$, $\rho_{11}(t) = \rho_{11}(0)$.
- **Off-diagonal elements** (coherences) decay exponentially:

$$
\rho_{01}(t) = \rho_{01}(0)\,\mathrm{e}^{-2\gamma t}, \qquad \rho_{10}(t) = \rho_{10}(0)\,\mathrm{e}^{-2\gamma t}
$$

**Bloch vector picture.** With $\hat{\rho} = (\hat{I} + \boldsymbol{r}\cdot\hat{\boldsymbol{\sigma}})/2$, the dephasing equation gives:

$$
\dot{r}_x = -2\gamma r_x, \quad \dot{r}_y = -2\gamma r_y, \quad \dot{r}_z = 0
$$

The Bloch vector **contracts toward the Z-axis** with timescale $T_2 = 1/(2\gamma)$, while its Z-component is preserved. This is the physics of decoherence: the qubit loses phase information (coherence) while retaining energy information (populations). After time $t \gg T_2$, any initial pure state $\vert\psi\rangle$ approaches the maximally mixed state $\hat{\rho} \to \rho_{00}\vert 0\rangle\langle 0\vert + \rho_{11}\vert 1\rangle\langle 1\vert$ with no off-diagonal elements.
:::


:::{admonition} Derivation: Quantum Jump (Monte Carlo) Unraveling of the Lindblad Equation
:class: dropdown information

The Lindblad equation $\dot{\hat{\rho}} = -\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] + \sum_k (\hat{L}_k\hat{\rho}\hat{L}_k^\dagger - \frac{1}{2}\hat{L}_k^\dagger\hat{L}_k\hat{\rho} - \frac{1}{2}\hat{\rho}\hat{L}_k^\dagger\hat{L}_k)$ can be written as an average over **stochastic quantum trajectories** — the quantum jump unraveling.

**No-jump evolution.** Define the non-Hermitian effective Hamiltonian:

$$
\hat{H}_{\mathrm{eff}} = \hat{H} - \frac{\mathrm{i}\hbar}{2}\sum_k \hat{L}_k^\dagger \hat{L}_k.
$$

Between jump events, a pure state $\vert\psi(t)\rangle$ evolves as

$$
\vert\tilde{\psi}(t+\mathrm{d}t)\rangle = \left(\hat{I} - \frac{\mathrm{i}}{\hbar}\hat{H}_{\mathrm{eff}}\,\mathrm{d}t\right)\vert\psi(t)\rangle.
$$

This **non-unitary** evolution causes $\Vert\vert\tilde{\psi}\rangle\Vert^2 = 1 - \delta p$ to decrease, where $\delta p = \mathrm{d}t \sum_k \langle\hat{L}_k^\dagger\hat{L}_k\rangle$ is the total jump probability in time $\mathrm{d}t$.

**Jump events.** With probability $\delta p_k = \mathrm{d}t\langle\hat{L}_k^\dagger\hat{L}_k\rangle$, a quantum jump of type $k$ occurs:

$$
\vert\psi\rangle \to \frac{\hat{L}_k\vert\psi\rangle}{\sqrt{\langle\hat{L}_k^\dagger\hat{L}_k\rangle}}.
$$

**Recovery of Lindblad.** The density matrix is the average over all trajectories. In time $\mathrm{d}t$:

$$
\hat{\rho}(t+\mathrm{d}t) = (1-\delta p)\frac{\vert\tilde{\psi}\rangle\langle\tilde{\psi}\vert}{1-\delta p} + \sum_k \delta p_k \frac{\hat{L}_k\vert\psi\rangle\langle\psi\vert\hat{L}_k^\dagger}{\langle\hat{L}_k^\dagger\hat{L}_k\rangle}.
$$

Expanding to first order in $\mathrm{d}t$ and averaging over the stochastic trajectories exactly recovers $\dot{\hat{\rho}} = \mathcal{L}[\hat{\rho}]$.

**Practical value.** For a large Hilbert space of dimension $d$, the density matrix requires $d^2$ numbers; each quantum trajectory requires only $d$ numbers. Simulating $N_{\mathrm{traj}}$ trajectories with $N_{\mathrm{traj}} \gg 1$ is far more efficient than integrating the full Lindblad equation, making the quantum jump method essential for simulating open many-body systems.
:::

### Summary

- The Lindblad equation $\dot{\hat{\rho}} = -\frac{\mathrm{i}}{\hbar}[\hat{H},\hat{\rho}] + \sum_k\gamma_k(\hat{L}_k\hat{\rho} \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k,\hat{\rho}\})$ is the most general Markovian master equation for open quantum systems.
- Lindblad operators $\hat{L}_k$ represent quantum jumps; rates $\gamma_k \geq 0$; the equation preserves trace, Hermiticity, and positivity (CPTP).
- Amplitude damping ($L = \hat{\sigma}_-$): decay to ground state. Phase damping ($L = \hat{\sigma}^z$): coherence decay without energy loss. Depolarizing ($L_\mu = \hat{\sigma}^\mu$): drives state to $I/2$.
- Steady states satisfy $\mathcal{L}(\hat{\rho}_\infty) = 0$; for thermal baths the steady state is the Boltzmann distribution $\mathrm{e}^{-\beta \hat{H}}/Z$.
- The quantum jump picture: Lindblad evolution is the average over stochastic trajectories where jumps occur at random times.
