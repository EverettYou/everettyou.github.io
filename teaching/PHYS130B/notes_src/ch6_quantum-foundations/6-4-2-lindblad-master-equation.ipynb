## Lecture Notes

### Overview

For an isolated system, the Schrödinger equation perfectly captures quantum evolution. But real systems interact with their surroundings. The Lindblad master equation describes how open systems evolve, accounting for both unitary dynamics and irreversible environmental effects. It is the most general Markovian (memoryless) evolution compatible with quantum mechanics.

The Lindblad equation decomposes open-system evolution into: (1) unitary evolution under the system Hamiltonian $\hat{H}$, and (2) dissipation described by jump operators $\{\hat{L}_k\}$ with rates $\{\gamma_k\}$. The equation preserves trace, Hermiticity, and positivity — necessary conditions for valid quantum dynamics. Every quantum channel (§6.3.3) is obtained from a Lindblad equation integrated over a fixed time interval.

:::{admonition} Physical Example: Spontaneously Emitting Atom
:class: note

Consider a two-level atom in an excited state $\vert 1\rangle$ coupled to the electromagnetic field. The excited state decays to the ground state $\vert 0\rangle$ by emitting a photon at rate $\Gamma_1$. Remarkably, the coherence between $\vert 0\rangle$ and $\vert 1\rangle$ outlives the excited-state population, decaying only half as fast. This asymmetry—coherences persist about twice as long as populations—makes spontaneous emission the simplest Lindblad process.
:::

:::{admonition} Discussion: Markovian Approximation
:class: dropdown tip

The Lindblad equation assumes uncorrelated system and environment at each instant—a Markovian approximation. But real environments can have memory. How do we know when Markovian is justified?

- The approximation breaks down if environment correlation time $\tau_E$ is comparable to the system timescale. For quantum optics, $\tau_E \sim 10^{-15}$ s while atomic evolution is $\sim 10^{-9}$ s—safely Markovian.
- Some systems show "non-Markovian revivals"—coherence temporarily returns. Does this invalidate the Lindblad equation, or is it simply outside its regime?
:::

### Lindblad Master Equation

For an open system coupled to an environment, the reduced density matrix $\hat{\rho}_S$ evolves as:

:::{admonition} Lindblad Master Equation (GKSL Form)
:class: important

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{\rho}] + \sum_k \gamma_k \left( \hat{L}_k \hat{\rho} \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k, \hat{\rho}\} \right)
$$ (eq-lindblad-equation)

where:
- $\hat{H}$ = system Hamiltonian (Hermitian), governing unitary evolution.
- $\hat{L}_k$ = **Lindblad operators** (jump operators); generally non-Hermitian.
- $\gamma_k \geq 0$ = **decay rates** (non-negative real numbers).
- $\{\hat{A}, \hat{B}\} = \hat{A}\hat{B} + \hat{B}\hat{A}$ = anticommutator.

This is the most general Markovian master equation compatible with complete positivity and trace preservation (Lindblad 1976; Kossakowski, Sudarshan — GKSL form).
:::

The Lindblad equation arises by tracing out the environment from unitary system+environment evolution:

:::{admonition} Microscopic Origin: Tracing Out the Environment
:class: dropdown note

The Lindblad form is what remains when an open system's environment is eliminated. The combined system+environment is closed, with total Hamiltonian

$$
\hat{H}_\text{total} = \hat{H}_S \otimes \hat{I}_E + \hat{I}_S \otimes \hat{H}_E + \hat{H}_\text{int}
$$

and evolves unitarily. The system alone is described by the reduced density matrix $\hat{\rho}_S = \operatorname{Tr}_E(\hat{\rho}_\text{total})$, obtained by tracing out the environment. Because the environment carries away information, $\hat{\rho}_S$ generically evolves non-unitarily.

When the environment correlation time is short compared to the system's evolution (the **Markovian approximation**), the environment is effectively memoryless — it resets between interactions, so the system's evolution depends only on its present state. The reduced dynamics then take the form

$$
\frac{\mathrm{d}\hat{\rho}_S}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}[\hat{H}_S, \hat{\rho}_S] + \mathcal{D}(\hat{\rho}_S)
$$

a unitary part plus a **dissipation superoperator** $\mathcal{D}$ encoding the environmental coupling.

The explicit construction of $\mathcal{D}$ from the microscopic coupling is beyond this section, but its *form* is fixed by two physical requirements alone: $\hat{\rho}_S$ must remain a valid density matrix, so $\mathcal{D}$ must preserve the trace ($\operatorname{Tr}(\hat{\rho}_S) = 1$) and complete positivity ($\hat{\rho}_S \geq 0$, even when the system is entangled with a spectator). The most general $\mathcal{D}$ meeting both constraints is exactly the jump-operator sum $\sum_k \gamma_k(\hat{L}_k\hat{\rho}_S \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k, \hat{\rho}_S\})$ — the GKSL theorem (Lindblad 1976; Gorini, Kossakowski, Sudarshan 1976). This is why the Lindblad equation is *the* Markovian master equation, not merely one model among many.
:::

### Properties of Lindblad Evolution

The Lindblad equation preserves all fundamental properties of quantum states:

**1. Trace preservation**: $\frac{\mathrm{d}}{\mathrm{d}t}\operatorname{Tr}(\hat{\rho}) = 0$, so $\operatorname{Tr}(\hat{\rho}(t)) = 1$ for all $t$.

**2. Hermiticity preservation**: If $\hat{\rho}(0) = \hat{\rho}(0)^\dagger$, then $\hat{\rho}(t) = \hat{\rho}(t)^\dagger$.

**3. Positivity**: If $\hat{\rho}(0) \geq 0$, then $\hat{\rho}(t) \geq 0$ for all $t$.

**4. CPTP map**: The time evolution is a completely positive trace-preserving (quantum) channel: $\hat{\rho}(t) = \sum_\mu \hat{K}_\mu \hat{\rho}(0) \hat{K}_\mu^\dagger$ with $\sum_\mu \hat{K}_\mu^\dagger \hat{K}_\mu = \hat{I}$.

**Gain and loss balance**: The $\hat{L}_k\hat{\rho} \hat{L}_k^\dagger$ term represents *gain* (probability inflow), while $-\frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k,\hat{\rho}\}$ represents *loss* (probability outflow). Together they preserve trace:

$$
\operatorname{Tr}\left(\hat{L}_k\hat{\rho} \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k,\hat{\rho}\}\right) = 0
$$

### Common Lindblad Operators

**1. Amplitude Damping: $\hat{L} = \sqrt{\Gamma_1}\,\hat{\sigma}_-$.**

**Physical process**: Excited state $\vert 1\rangle$ decays to ground state $\vert 0\rangle$ (spontaneous emission) at rate $\Gamma_1 = 1/T_1$. For the two-level atom, the Lindblad operator $\hat{L} = \sqrt{\Gamma_1}\,\vert 0\rangle\langle 1\vert$ describes spontaneous emission at rate $\Gamma_1$.

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = \Gamma_1\!\left(\hat{\sigma}_-\hat{\rho}\hat{\sigma}_+ - \frac{1}{2}\{\hat{\sigma}_+\hat{\sigma}_-,\hat{\rho}\}\right)
$$ (eq-lindblad-amp-damping)

**Effect**: $\dot{\rho}_{11} = -\Gamma_1\rho_{11}$ (decay), $\dot{\rho}_{00} = +\Gamma_1\rho_{11}$ (gain), $\dot{\rho}_{01} = -\frac{\Gamma_1}{2}\rho_{01}$ (coherences decay at half rate).

**2. Phase Damping: $\hat{L} = \sqrt{\Gamma_\varphi}\,\hat{Z}$.**

**Physical process**: Random phase fluctuations (dephasing) with no energy loss; the resulting coherence decays at rate $2\Gamma_\varphi$, so the pure dephasing time is $T_\varphi = 1/(2\Gamma_\varphi)$.

$$
\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = \Gamma_\varphi\!\left(\hat{Z}\hat{\rho}\hat{Z} - \hat{\rho}\right)
$$ (eq-lindblad-phase-damping)

**Effect**: Only off-diagonal terms decay: $\rho_{01}(t) = \mathrm{e}^{-2\Gamma_\varphi t}\rho_{01}(0)$. Diagonal terms unchanged—this is the microscopic realization of the dephasing channel (§6.3.3, §6.4.1).

**3. Depolarizing: $\hat{L}_1 = \sqrt{\Gamma_d/3}\,\hat{X}$, $\hat{L}_2 = \sqrt{\Gamma_d/3}\,\hat{Y}$, $\hat{L}_3 = \sqrt{\Gamma_d/3}\,\hat{Z}$.**

**Physical process**: Symmetric noise at rate $\Gamma_d$.

**Effect**: Drives any state toward the maximally mixed state $\hat{I}/2$, with the deviation $\hat{\rho} - \hat{I}/2$ decaying as $\mathrm{e}^{-4\Gamma_d t/3}$ (e-folding time $3/(4\Gamma_d)$).

### Steady States

A **steady state** $\hat{\rho}_\infty$ satisfies $\mathcal{L}(\hat{\rho}_\infty) = 0$.

:::{admonition} Example: Amplitude Damping
:class: dropdown example

**Problem.** Find the steady state for amplitude damping $\hat{L} = \sqrt{\Gamma_1}\hat{\sigma}_-$.

**Solution.** From $\dot{\rho}_{11} = -\Gamma_1\rho_{11} = 0$ we get $\rho_{11}^\infty = 0$, so $\rho_{00}^\infty = 1$. The steady state is $\hat{\rho}_\infty = \vert 0\rangle\langle 0\vert$.
:::

:::{admonition} Example: Thermal Steady State
:class: dropdown example

**Problem.** For an atom in a thermal bath at temperature $T$, include decay ($\hat{L}_1 = \sqrt{\Gamma_1\bar{n}_\text{th}+\Gamma_1}\,\hat{\sigma}_-$) and absorption ($\hat{L}_2 = \sqrt{\Gamma_1\bar{n}_\text{th}}\,\hat{\sigma}_+$). Find the steady state.

**Solution.** The two jump operators move population in opposite directions: $\hat{L}_2$ (absorption) pumps $\vert 0\rangle \to \vert 1\rangle$ at rate $\Gamma_\uparrow = \Gamma_1\bar{n}_\text{th}$, while $\hat{L}_1$ (decay) drains $\vert 1\rangle \to \vert 0\rangle$ at rate $\Gamma_\downarrow = \Gamma_1\bar{n}_\text{th}+\Gamma_1 = \Gamma_1(\bar{n}_\text{th}+1)$. Reading off the gain and loss terms as for amplitude damping, the excited-state population obeys

$$
\dot{\rho}_{11} = \Gamma_\uparrow\,\rho_{00} - \Gamma_\downarrow\,\rho_{11}
$$

At steady state $\dot{\rho}_{11} = 0$, so the upward and downward population flows balance, which fixes the population ratio:

$$
\begin{split}
\Gamma_\uparrow\,\rho_{00}^\infty &= \Gamma_\downarrow\,\rho_{11}^\infty \\
\frac{\rho_{11}^\infty}{\rho_{00}^\infty} &= \frac{\Gamma_\uparrow}{\Gamma_\downarrow} = \frac{\Gamma_1\bar{n}_\text{th}}{\Gamma_1(\bar{n}_\text{th}+1)} = \frac{\bar{n}_\text{th}}{\bar{n}_\text{th}+1}
\end{split}
$$

Inserting the Bose-Einstein occupation $\bar{n}_\text{th} = 1/(\mathrm{e}^{\hbar\omega/k_BT}-1)$, so that $\bar{n}_\text{th}+1 = \mathrm{e}^{\hbar\omega/k_BT}/(\mathrm{e}^{\hbar\omega/k_BT}-1)$, the ratio collapses to the Boltzmann factor:

$$
\frac{\rho_{11}^\infty}{\rho_{00}^\infty} = \frac{\bar{n}_\text{th}}{\bar{n}_\text{th}+1} = \mathrm{e}^{-\hbar\omega/k_BT}
$$

This is exactly the population ratio of the thermal state $\hat{\rho}_\infty = \mathrm{e}^{-\beta\hat{H}}/Z$ with $\hat{H} = \hbar\omega\,\vert 1\rangle\langle 1\vert$ and $\beta = 1/k_BT$ (§6.1.2): dissipation in contact with a thermal bath drives the atom toward thermal equilibrium.
:::

### Quantum Jump Picture

The Lindblad equation averages over an ensemble of quantum trajectories. Each trajectory evolves stochastically:

1. **Between jumps**: The state evolves under the *non-Hermitian* effective Hamiltonian

$$
\hat{H}_\text{eff} = \hat{H} - \frac{\mathrm{i}\hbar}{2}\sum_k\gamma_k \hat{L}_k^\dagger \hat{L}_k
$$ (eq-lindblad-heff)

   The imaginary part causes the norm to decay, tracking the "no-jump" probability.

2. **At random times** (Poisson process), a **quantum jump** $\hat{L}_k$ occurs: $\hat{\rho} \to \hat{L}_k\hat{\rho} \hat{L}_k^\dagger/p_k$ with probability $p_k \propto \gamma_k\,\operatorname{Tr}(\hat{L}_k^\dagger \hat{L}_k\hat{\rho})\,\mathrm{d}t$.

3. **Averaging** over all trajectories recovers the Lindblad master equation.

**Physical interpretation**: Each quantum jump corresponds to a physical event—photon emission, phonon absorption, spin flip. The non-Hermitian Hamiltonian reduces the state's norm between jumps, tracking the probability that no jump has occurred. When a jump happens, the state is projected and renormalized. Averaging over many trajectories with different jump sequences recovers the Lindblad master equation.

### Summary

- **Lindblad master equation:** $\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = -\frac{\mathrm{i}}{\hbar}[\hat{H}, \hat{\rho}] + \sum_k \gamma_k (\hat{L}_k \hat{\rho} \hat{L}_k^\dagger - \frac{1}{2}\{\hat{L}_k^\dagger \hat{L}_k, \hat{\rho}\})$ (GKSL); describes Markovian open-system dynamics.

- **Jump operators $\hat{L}_k$:** Represent decoherence and dissipation; decay rates $\gamma_k \geq 0$. Amplitude damping, dephasing, and decay are modeled by specific choices.

- **Steady state:** Long-time limit $\hat{\rho}(t \to \infty) = \hat{\rho}_\infty$ satisfies $\frac{\mathrm{d}\hat{\rho}}{\mathrm{d}t} = 0$. Describes equilibrium or thermal state.

- **Positivity and CPTP:** Lindblad form guarantees complete positivity and trace preservation for any $\hat{L}_k$ and $\gamma_k \geq 0$.

- **Quantum jump picture:** The Lindblad equation is the ensemble average of stochastic quantum trajectories — smooth evolution under the non-Hermitian effective Hamiltonian $\hat{H}_\text{eff}$, punctuated by random jumps $\hat{L}_k$ at Poisson-distributed times.
