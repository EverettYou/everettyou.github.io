## Lecture Notes

### Overview

The maximum entropy principle (§6.1.2) immediately yields the quantum distributions governing bosons and fermions. By applying the constraint of fixed average energy to a single mode, we derive the Bose-Einstein and Fermi-Dirac distributions from first principles. A single mode is sufficient to capture the essential physics: the statistics constrain thermodynamics.

### Bose-Einstein Distribution

#### Single Bosonic Mode

Consider a single bosonic mode with Hamiltonian:

$$
\hat{H} = \varepsilon \hat{b}^\dagger \hat{b}
$$

The eigenstates are Fock states $\vert n \rangle$ with $n = 0, 1, 2, \ldots$ (unbounded occupation), and eigenvalues $E_n = n\varepsilon$.

At thermal equilibrium, the system obeys the maximum entropy principle: the thermal (Gibbs) state is:

$$
\hat{\rho} = \frac{\mathrm{e}^{-\beta \hat{H}}}{Z}
$$

where $\beta = 1/(k_B T)$ and $Z = \text{Tr}(\mathrm{e}^{-\beta \hat{H}})$ is the partition function.

#### Partition Function

The partition function is:

$$
Z = \sum_{n=0}^{\infty} \mathrm{e}^{-\beta n \varepsilon} = \sum_{n=0}^{\infty} \left(\mathrm{e}^{-\beta \varepsilon}\right)^n

$$

This is a geometric series with ratio $r = \mathrm{e}^{-\beta \varepsilon} < 1$:

$$
Z = \frac{1}{1 - \mathrm{e}^{-\beta \varepsilon}}

$$ (eq-qstat-boson-Z)

:::{admonition} Derivation: Geometric Series for Partition Function
:class: dropdown information

For $|r| < 1$, the geometric series is:

$$
\sum_{n=0}^{\infty} r^n = \frac{1}{1-r}

$$

With $r = \mathrm{e}^{-\beta\varepsilon}$, we have $r < 1$ (since $\beta\varepsilon > 0$), so:

$$
Z = \frac{1}{1 - \mathrm{e}^{-\beta\varepsilon}}

$$

This partition function is valid for all $T > 0$ (no phase transition, no Bose-Einstein condensation in a single mode).

:::

#### Average Occupation Number

The average occupation is the thermal expectation value:

$$
\langle n \rangle = \text{Tr}(\hat{\rho} \hat{n}) = \frac{\text{Tr}(\mathrm{e}^{-\beta \hat{H}} \hat{n})}{Z}

$$

where $\hat{n} = \hat{b}^\dagger \hat{b}$ is the number operator.

Expanding in the Fock basis:

$$
\text{Tr}(\mathrm{e}^{-\beta \hat{H}} \hat{n}) = \sum_{n=0}^{\infty} n \mathrm{e}^{-\beta n \varepsilon} = \mathrm{e}^{-\beta\varepsilon} \frac{\mathrm{d}}{\mathrm{d}(\mathrm{e}^{-\beta\varepsilon})} Z

$$

Taking the derivative:

$$
\langle n \rangle = -\frac{1}{Z}\frac{\mathrm{d}Z}{\mathrm{d}\beta}

$$

After computation:


:::{admonition} Derivation: Bose-Einstein Average Occupation
:class: dropdown information

The average occupation is:

$$
\langle n\rangle = \frac{1}{Z}\sum_{n=0}^\infty n\,\mathrm{e}^{-\beta n\varepsilon}
$$

**Step 1: Rewrite as a derivative.** Differentiate $Z = \sum_{n=0}^\infty \mathrm{e}^{-\beta n\varepsilon}$ with respect to $\beta$:

$$
\frac{\partial Z}{\partial\beta} = -\varepsilon\sum_{n=0}^\infty n\,\mathrm{e}^{-\beta n\varepsilon}
$$

So $\sum_{n=0}^\infty n\,\mathrm{e}^{-\beta n\varepsilon} = -\frac{1}{\varepsilon}\frac{\partial Z}{\partial\beta}$, giving:

$$
\langle n\rangle = -\frac{1}{Z\varepsilon}\frac{\partial Z}{\partial\beta} = -\frac{1}{\varepsilon}\frac{\partial \ln Z}{\partial\beta}
$$

**Step 2: Differentiate $\ln Z$.** From $Z = 1/(1-\mathrm{e}^{-\beta\varepsilon})$:

$$
\ln Z = -\ln(1 - \mathrm{e}^{-\beta\varepsilon})
$$

$$
\frac{\partial \ln Z}{\partial\beta} = -\frac{\varepsilon\,\mathrm{e}^{-\beta\varepsilon}}{1-\mathrm{e}^{-\beta\varepsilon}}
$$

**Step 3: Combine.** Substituting:

$$
\langle n\rangle = -\frac{1}{\varepsilon}\cdot\left(-\frac{\varepsilon\,\mathrm{e}^{-\beta\varepsilon}}{1-\mathrm{e}^{-\beta\varepsilon}}\right) = \frac{\mathrm{e}^{-\beta\varepsilon}}{1-\mathrm{e}^{-\beta\varepsilon}} = \frac{1}{\mathrm{e}^{\beta\varepsilon}-1}
$$

Multiplying numerator and denominator by $\mathrm{e}^{\beta\varepsilon}$ gives the Bose-Einstein distribution. $\square$
:::

:::{admonition} Bose-Einstein Distribution
:class: important

The average occupation number of a single bosonic mode is:

$$
\langle n \rangle = \frac{1}{\mathrm{e}^{\beta\varepsilon} - 1}

$$ (eq-qstat-bose-einstein)

where $\beta = 1/(k_B T)$ and $\varepsilon$ is the single-particle energy.

This is the **Bose-Einstein (BE) distribution**. Bosons can accumulate in the same state (no limit on $n$).
:::

#### Physical Meaning

- **High temperature** ($\beta\varepsilon \ll 1$): $\langle n \rangle \approx 1/(\beta\varepsilon) \to \infty$ — many particles populate the mode.
- **Low temperature** ($\beta\varepsilon \gg 1$): $\langle n \rangle \approx \mathrm{e}^{-\beta\varepsilon} \to 0$ — occupation suppressed, but can be any number.
- **No Pauli exclusion**: Arbitrary occupation is allowed. Bosons "bunch"—they prefer to occupy the same state.

### Fermi-Dirac Distribution

#### Single Fermionic Mode

Consider a single fermionic mode with Hamiltonian:

$$
\hat{H} = \varepsilon \hat{c}^\dagger \hat{c}

$$

The crucial difference: fermionic anticommutation relations impose Pauli exclusion. The operator $\hat{c}^\dagger \hat{c}$ has only two eigenvalues: $n = 0$ (empty) and $n = 1$ (occupied), with energies $E_0 = 0$ and $E_1 = \varepsilon$.

#### Partition Function

The partition function has only two terms:

$$
Z = \mathrm{e}^{-\beta \cdot 0} + \mathrm{e}^{-\beta\varepsilon} = 1 + \mathrm{e}^{-\beta\varepsilon}

$$ (eq-qstat-fermion-Z)

Simple and elegant.

#### Average Occupation Number

$$
\langle n \rangle = \frac{0 \cdot 1 + 1 \cdot \mathrm{e}^{-\beta\varepsilon}}{Z} = \frac{\mathrm{e}^{-\beta\varepsilon}}{1 + \mathrm{e}^{-\beta\varepsilon}}

$$

Dividing numerator and denominator by $\mathrm{e}^{-\beta\varepsilon}$:

:::{admonition} Fermi-Dirac Distribution
:class: important

The average occupation number of a single fermionic mode is:

$$
\langle n \rangle = \frac{1}{\mathrm{e}^{\beta\varepsilon} + 1}

$$ (eq-qstat-fermi-dirac)

This is the **Fermi-Dirac (FD) distribution**. At most one fermion occupies a single state (Pauli exclusion principle).
:::

#### Physical Meaning

- **High temperature** ($\beta\varepsilon \ll 1$): $\langle n \rangle \approx 1/2$ — the state is equally likely occupied or empty.
- **Low temperature** ($\beta\varepsilon \gg 1$): $\langle n \rangle \approx \mathrm{e}^{-\beta\varepsilon} \to 0$ — state mostly empty, but exponentially suppressed.
- **Pauli exclusion**: Occupation is always $0 \leq \langle n \rangle \leq 1$. Fermions "exclude"—no two can occupy the same state.

### Bosons vs Fermions: Side-by-Side Comparison

| Quantity | Bosons (BE) | Fermions (FD) |
|----------|-------------|---------------|
| Max occupation | Unbounded | 1 (Pauli) |
| Partition function | $Z = 1/(1 - \mathrm{e}^{-\beta\varepsilon})$ | $Z = 1 + \mathrm{e}^{-\beta\varepsilon}$ |
| $\langle n \rangle$ | $1/(\mathrm{e}^{\beta\varepsilon} - 1)$ | $1/(\mathrm{e}^{\beta\varepsilon} + 1)$ |
| Denominator | $(\cdots - 1)$ | $(\cdots + 1)$ |
| Low $T$ | $\langle n \rangle \approx \mathrm{e}^{-\beta\varepsilon}$ | $\langle n \rangle \approx \mathrm{e}^{-\beta\varepsilon}$ |
| High $T$ | $\langle n \rangle \to \infty$ | $\langle n \rangle \to 1/2$ |
| Physics | Bunching; condensation | Exclusion; saturation |

:::{admonition} Discussion
:class: dropdown tip

The $\pm 1$ in the denominators encodes a profound difference. For bosons, as temperature drops, can occupation diverge? For fermions, why is high-temperature occupation pinned at $1/2$? How does this connect to exchange symmetry and statistics from Chapter 2?
:::

**Connection to Chapter 2**: The statistics (exchange symmetry of the wavefunction) directly constrains the thermodynamic behavior. Bosons, with symmetric wavefunctions, can pile up. Fermions, with antisymmetric wavefunctions, must exclude.


:::{admonition} Derivation: Chemical Potential and Grand Canonical Ensemble
:class: dropdown information

In the grand canonical ensemble (fixed $T$ and $\mu$ rather than fixed $N$), the thermal state is $\hat{\rho} = \mathrm{e}^{-\beta(\hat{H}-\mu\hat{N})}/\mathcal{Z}$ where $\mathcal{Z} = \operatorname{Tr}(\mathrm{e}^{-\beta(\hat{H}-\mu\hat{N})})$.

**Single bosonic mode:** Replace $\varepsilon \to \varepsilon - \mu$ in all formulas. The grand partition function:

$$
\mathcal{Z}_{\text{bose}} = \frac{1}{1 - \mathrm{e}^{-\beta(\varepsilon-\mu)}}, \quad (\mu < \varepsilon \text{ required for convergence})
$$

gives the full Bose-Einstein distribution:

$$
\langle n\rangle_{\text{BE}} = \frac{1}{\mathrm{e}^{\beta(\varepsilon-\mu)}-1}
$$

**Single fermionic mode:** $\mathcal{Z}_{\text{fermi}} = 1 + \mathrm{e}^{-\beta(\varepsilon-\mu)}$, giving:

$$
\langle n\rangle_{\text{FD}} = \frac{1}{\mathrm{e}^{\beta(\varepsilon-\mu)}+1}
$$

**Chemical potential $\mu$** is determined by fixing the total particle number $N = \sum_{\text{modes}} \langle n\rangle$. At $T=0$ for fermions, $\mu$ equals the **Fermi energy** $E_F$: all modes with $\varepsilon < E_F$ are filled ($\langle n\rangle = 1$), all with $\varepsilon > E_F$ are empty ($\langle n\rangle = 0$).
:::

:::{admonition} Derivation: Planck's Radiation Law from the Bose-Einstein Distribution
:class: dropdown information

Photons are bosons with **zero chemical potential** ($\mu = 0$), since their number is not conserved — the walls of a blackbody cavity freely absorb and emit them. The average number of photons in a mode of frequency $\omega$ is therefore:

$$
\langle n_\omega \rangle = \frac{1}{\mathrm{e}^{\hbar\omega/k_BT} - 1}
$$

**Step 1: Density of states for photons.** In a box of volume $V$, the number of electromagnetic modes with wavevector magnitude in $[k, k+\mathrm{d}k]$ is (with two polarizations):

$$
g(k)\,\mathrm{d}k = \frac{2 \cdot 4\pi k^2\,\mathrm{d}k}{(2\pi)^3} \cdot V = \frac{V k^2}{{\pi^2}}\,\mathrm{d}k
$$

Converting to angular frequency using $\omega = ck$ ($c$ = speed of light), $\mathrm{d}k = \mathrm{d}\omega/c$:

$$
g(\omega)\,\mathrm{d}\omega = \frac{V\omega^2}{\pi^2 c^3}\,\mathrm{d}\omega
$$

**Step 2: Energy density.** The total electromagnetic energy in $[\omega, \omega+\mathrm{d}\omega]$ is (energy per photon $\hbar\omega$ times occupation times density of states):

$$
u(\omega)\,\mathrm{d}\omega = \frac{\hbar\omega}{V}\cdot g(\omega)\cdot\langle n_\omega\rangle\,\mathrm{d}\omega = \frac{\hbar\omega^3}{\pi^2 c^3}\cdot\frac{1}{\mathrm{e}^{\hbar\omega/k_BT}-1}\,\mathrm{d}\omega
$$

This is **Planck's radiation law** for the spectral energy density $u(\omega)$.

**Step 3: Stefan-Boltzmann law.** Integrating over all frequencies using the substitution $x = \hbar\omega/k_BT$:

$$
u_{\text{total}} = \int_0^\infty u(\omega)\,\mathrm{d}\omega = \frac{(k_BT)^4}{\pi^2 c^3 \hbar^3}\int_0^\infty \frac{x^3}{\mathrm{e}^x - 1}\,\mathrm{d}x = \frac{\pi^2}{15}\cdot\frac{(k_BT)^4}{\hbar^3 c^3}
$$

where $\int_0^\infty x^3/(\mathrm{e}^x-1)\,\mathrm{d}x = \pi^4/15$. The energy density scales as $T^4$ — the **Stefan-Boltzmann law**.

**Classical (Rayleigh-Jeans) limit.** For $\hbar\omega \ll k_BT$, expanding $\mathrm{e}^x \approx 1 + x$:

$$
u(\omega) \approx \frac{\omega^2}{\pi^2 c^3}\cdot k_BT
$$

This is the **Rayleigh-Jeans law** — the classical result that diverges as $\omega^2$ ("ultraviolet catastrophe"). Planck's quantum formula avoids this: the BE factor $1/(\mathrm{e}^x-1)$ exponentially suppresses high-frequency modes.

**Historical significance.** Planck's 1900 result was the first use of energy quantization and the direct origin of quantum mechanics. The $\pm 1$ in the denominator — the signature of Bose-Einstein statistics — resolves the ultraviolet catastrophe.
:::

:::{admonition} Derivation: Fermi-Dirac Distribution from the Grand Canonical Ensemble
:class: dropdown information

The Fermi-Dirac distribution can be derived cleanly by incorporating a chemical potential $\mu$ into the statistical ensemble.

**Step 1: Grand canonical partition function.**

To allow particle number to fluctuate (at fixed average), replace the Boltzmann weight $\mathrm{e}^{-\beta\varepsilon n}$ with the grand canonical weight $\mathrm{e}^{-\beta(\varepsilon - \mu)n}$, where $\mu$ is the chemical potential:

$$
Z_\text{GC} = \sum_{n=0}^{N_\text{max}} \mathrm{e}^{-\beta(\varepsilon-\mu)n}
$$

For **fermions**, the Pauli exclusion principle limits $n \in \{0,1\}$, so:

$$
Z_\text{GC}^\text{FD} = 1 + \mathrm{e}^{-\beta(\varepsilon-\mu)}
$$

For **bosons**, $n = 0,1,2,\ldots$ and the geometric series gives $Z_\text{GC}^\text{BE} = 1/(1 - \mathrm{e}^{-\beta(\varepsilon-\mu)})$ (requiring $\varepsilon > \mu$).

**Step 2: Average occupation.**

Differentiate $-\partial_{\beta(\varepsilon-\mu)} \ln Z_\text{GC}$ to get $\langle n \rangle$:

$$
\langle n \rangle = -\frac{\partial \ln Z_\text{GC}}{\partial[\beta(\varepsilon-\mu)]}
$$

For fermions:

$$
\langle n \rangle_\text{FD} = -\frac{\partial}{\partial[\beta(\varepsilon-\mu)]} \ln\!\left(1 + \mathrm{e}^{-\beta(\varepsilon-\mu)}\right) = \frac{\mathrm{e}^{-\beta(\varepsilon-\mu)}}{1+\mathrm{e}^{-\beta(\varepsilon-\mu)}} = \frac{1}{\mathrm{e}^{\beta(\varepsilon-\mu)}+1}
$$

**Step 3: Low-temperature (Fermi energy) limit.**

At $T \to 0$ ($\beta \to \infty$):

$$
\langle n \rangle_\text{FD} = \frac{1}{\mathrm{e}^{\beta(\varepsilon-\mu)}+1} \;\longrightarrow\; \begin{cases} 1 & \text{if } \varepsilon < \mu \\\ 1/2 & \text{if } \varepsilon = \mu \\\ 0 & \text{if } \varepsilon > \mu \end{cases}
$$

All states below $\mu$ are fully occupied; all above are empty. This defines the **Fermi energy** $E_F = \mu(T=0)$: the chemical potential at zero temperature.

**Step 4: Fixing $\mu$ from total particle number.**

For a system with $d$ modes at energies $\varepsilon_1 \leq \varepsilon_2 \leq \cdots$, the chemical potential is determined self-consistently:

$$
N = \sum_k \langle n_k \rangle = \sum_k \frac{1}{\mathrm{e}^{\beta(\varepsilon_k-\mu)}+1}
$$

At $T=0$, this equation simply fills the $N$ lowest energy states: $E_F = \varepsilon_N$ (the energy of the highest occupied orbital). As $T$ increases, $\mu(T)$ decreases from $E_F$ to keep $N$ fixed.

**Summary:** The Fermi-Dirac distribution $\langle n \rangle = 1/(\mathrm{e}^{\beta(\varepsilon-\mu)}+1)$ emerges from the grand canonical ensemble with Pauli exclusion. The $+1$ in the denominator (vs. $-1$ for bosons) encodes the Pauli exclusion principle at the thermodynamic level.
:::

### The Classical Limit

When temperature is high or the energy $\varepsilon$ is low, the quantum effects become negligible and both distributions should reduce to classical statistical mechanics.

#### Condition: $\mathrm{e}^{\beta\varepsilon} \gg 1$

In this regime:

**Bosons:**

$$
\langle n \rangle_{\text{BE}} = \frac{1}{\mathrm{e}^{\beta\varepsilon} - 1} \approx \frac{1}{\mathrm{e}^{\beta\varepsilon}} = \mathrm{e}^{-\beta\varepsilon}

$$

**Fermions:**

$$
\langle n \rangle_{\text{FD}} = \frac{1}{\mathrm{e}^{\beta\varepsilon} + 1} \approx \frac{1}{\mathrm{e}^{\beta\varepsilon}} = \mathrm{e}^{-\beta\varepsilon}

$$

#### Boltzmann Distribution

Both reduce to:

$$
\langle n \rangle = \mathrm{e}^{-\beta\varepsilon}

$$ (eq-qstat-boltzmann)

This is the **Boltzmann distribution**: the average energy fraction of a classical harmonic oscillator or any mode at energy $\varepsilon$ in thermal equilibrium.

**Physical Interpretation**: At high temperature, the quantum statistical nature (BE vs FD) becomes invisible. The system behaves classically. This is the **correspondence principle**: quantum and classical statistics agree when quantum effects are negligible ($\hbar \omega \ll k_B T$).

### Summary

- **Bose-Einstein distribution**: $\langle n \rangle = 1/(\mathrm{e}^{\beta\varepsilon} - 1)$ from a single bosonic mode with unbounded occupation.
- **Fermi-Dirac distribution**: $\langle n \rangle = 1/(\mathrm{e}^{\beta\varepsilon} + 1)$ from a single fermionic mode with $n \in \{0,1\}$ only.
- **The $\pm 1$ difference**: Encodes exchange symmetry. Bosons bunch; fermions exclude.
- **Partition functions**: Geometric series (bosons) vs two-term sum (fermions), yet both encode the same physics.
- **Classical limit**: When $\mathrm{e}^{\beta\varepsilon} \gg 1$, both reduce to the Boltzmann distribution $\mathrm{e}^{-\beta\varepsilon}$.
- **Connection to Chapter 2**: Identical particle statistics (exchange symmetry) determine thermodynamic behavior. Quantum statistics and quantum mechanics are inseparable.
