## Lecture Notes

### Overview

Measurement in quantum mechanics is not a passive observation—it is a projective act that forces the system into one of several possible states. For any Hermitian observable, the measurement outcome is an eigenvalue, the probability depends on the initial state's alignment with the corresponding eigenspace (Born rule), and the post-measurement state is the normalized projection onto that eigenspace. This section develops the mathematical framework and explores key consequences: repeated measurements give the same outcome with certainty, frequent measurements can suppress evolution (quantum Zeno effect), and commuting observables can be measured simultaneously.

### Spectral Decomposition and Hermitian Operators

:::{admonition} Definition: Spectral Decomposition
:class: important

For a Hermitian operator $\hat{A}$ (observable), the **spectral theorem** states:

$$
\hat{A} = \sum_i a_i \hat{P}_i
$$ (eq-projmeas-spectral)

where:
- $a_i$ = eigenvalues (real, since $\hat{A}$ is Hermitian)
- $\hat{P}_i$ = projector onto eigenspace of $a_i$
- $\hat{P}_i$ are **orthogonal**: $\hat{P}_i \hat{P}_j = \delta_{ij} \hat{P}_i$
- $\hat{P}_i$ are **complete**: $\sum_i \hat{P}_i = \hat{I}$ (sum to identity)
- $\hat{P}_i^2 = \hat{P}_i$ (idempotent)

If eigenvalue $a_i$ is **degenerate** (multiplicity $> 1$), projector $\hat{P}_i$ projects onto a multi-dimensional subspace.
:::

:::{admonition} Example: Qubit Spin Observable
:class: example

For $\hat{Z} = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ (Pauli-z operator):

- Eigenvalues: $\lambda_1 = +1, \lambda_2 = -1$.
- Eigenvectors: $\vert 0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \vert 1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.
- Projectors:

$$
\begin{split}
\hat{P}_+ &= \vert 0\rangle\langle 0\vert = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}, \\
\hat{P}_- &= \vert 1\rangle\langle 1\vert = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix}
\end{split}
$$

- Verification: $\hat{P}_+ + \hat{P}_- = \hat{I}$ ✓, $\hat{P}_+^2 = \hat{P}_+$ ✓, $\hat{P}_+ \hat{P}_- = 0$ ✓.
- Spectral form: $\hat{Z} = (+1) \hat{P}_+ + (-1) \hat{P}_-$.

:::

### The Measurement Postulate

Recall the **measurement postulate** from §1.2.1: For a system in state $\hat{\rho}$ measured for observable $\hat{A} = \sum_i a_i \hat{P}_i$, the outcome is eigenvalue $a_i$ with probability $p_i = \operatorname{Tr}(\hat{P}_i \hat{\rho})$, and the state collapses to $\hat{\rho}' = \hat{P}_i \hat{\rho} \hat{P}_i / \operatorname{Tr}(\hat{P}_i \hat{\rho})$.

Here we explore the mathematical structure more carefully, generalizations, and physical consequences.

:::{admonition} Measurement Postulate: Probability and Collapse
:class: important

**Outcome**: Measurement yields eigenvalue $a_i$ with **probability**:

$$
p_i = \operatorname{Tr}(\hat{P}_i \hat{\rho})
$$ (eq-projmeas-probability)

**Post-measurement state** (state collapse):

$$
\hat{\rho}' = \frac{\hat{P}_i \hat{\rho} \hat{P}_i}{\operatorname{Tr}(\hat{P}_i \hat{\rho})} = \frac{\hat{P}_i \hat{\rho} \hat{P}_i}{p_i}
$$ (eq-projmeas-collapse)

The state is **projected** onto the eigenspace corresponding to the measured eigenvalue.
:::


**Normalization Check.**

$$
\operatorname{Tr}(\hat{\rho}') = \frac{\operatorname{Tr}(\hat{P}_i \hat{\rho} \hat{P}_i)}{p_i} = \frac{p_i}{p_i} = 1
$$

since $\hat{P}_i^2 = \hat{P}_i$ and the trace is cyclic.

:::{admonition} Born Rule: Expectation Value
:class: important

The average outcome of repeated measurements is:

$$
\langle \hat{A} \rangle = \sum_i a_i p_i = \sum_i a_i \operatorname{Tr}(\hat{P}_i \hat{\rho}) = \operatorname{Tr}(\hat{A} \hat{\rho})
$$ (eq-projmeas-expectation)

This is the **Born rule** in density matrix form.
:::

:::{admonition} Measurement of Pure States
:class: note

For a **pure state** $\hat{\rho} = \vert\psi\rangle\langle\psi\vert$:

**Probability**:

$$
p_i = \operatorname{Tr}(\hat{P}_i \vert\psi\rangle\langle\psi\vert) = \langle\psi\vert \hat{P}_i \vert\psi\rangle
$$ (eq-projmeas-pure-prob)

If $\vert\psi\rangle = \sum_j c_j \vert u_j^{(i)}\rangle$ where $\vert u_j^{(i)}\rangle$ are eigenstates of eigenvalue $a_i$ (with multiplicity $d_i$):

$$
p_i = \sum_j \vert c_j\vert^2
$$

**Post-measurement state**:

$$
\vert\psi'\rangle = \frac{\hat{P}_i \vert\psi\rangle}{\sqrt{p_i}}
$$ (eq-projmeas-pure-collapse)

The state is **normalized projection** onto eigenspace $i$.
:::

:::{admonition} Example: Measuring $\hat{Z}$
:class: example

Initial state: $\vert\psi\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$ (superposition).

Measure $\hat{Z}$:
- **Outcome** $+1$ ($\vert 0\rangle$):
  - Probability: $p_+ = \vert\frac{1}{\sqrt{2}}\vert^2 = 1/2$.
  - Post-measurement: $\vert\psi'\rangle = \vert 0\rangle$ (definite $\vert 0\rangle$).

- **Outcome** $-1$ ($\vert 1\rangle$):
  - Probability: $p_- = \vert\frac{1}{\sqrt{2}}\vert^2 = 1/2$.
  - Post-measurement: $\vert\psi'\rangle = \vert 1\rangle$ (definite $\vert 1\rangle$).

After measurement, the superposition is **destroyed**; the state is definite.

:::

### Repeated Measurements

**Theorem**: If you measure observable $\hat{A}$ and obtain outcome $a_i$, then immediately measure $\hat{A}$ again, you get the **same outcome** $a_i$ with **certainty** (probability 1).

**Proof**: After first measurement, state is $\hat{\rho}' = \hat{P}_i \hat{\rho} \hat{P}_i / p_i$, which is **supported** on eigenspace of $a_i$.

Second measurement probability:

$$
p_i^{(2)} = \operatorname{Tr}(\hat{P}_i \hat{\rho}') = \operatorname{Tr}\left(\hat{P}_i \frac{\hat{P}_i \hat{\rho} \hat{P}_i}{p_i}\right) = \frac{\operatorname{Tr}(\hat{P}_i \hat{P}_i \hat{\rho} \hat{P}_i)}{p_i} = \frac{\operatorname{Tr}(\hat{P}_i \hat{\rho} \hat{P}_i)}{p_i} = \frac{p_i}{p_i} = 1
$$

Here $\hat{P}_i \hat{P}_i = \hat{P}_i$ by idempotence, and the final equality uses the cyclic invariance of the trace: $\operatorname{Tr}(\hat{P}_i \hat{\rho} \hat{P}_i) = \operatorname{Tr}(\hat{P}_i \hat{P}_i \hat{\rho}) = \operatorname{Tr}(\hat{P}_i \hat{\rho}) = p_i$.

**Consequence**: Measurement outcome is perfectly deterministic on repeated measurement; the result **doesn't change**.

### The Quantum Zeno Effect

**Setup**: System starts in state $\vert\psi_0\rangle$ that would evolve to $\vert\psi(t)\rangle$ under unitary evolution (e.g., decay).

**Frequent projective measurements**: Measure at times $0, \Delta t, 2\Delta t, \ldots$ whether the system is still in $\vert\psi_0\rangle$ (projector $\hat{P} = \vert\psi_0\rangle\langle\psi_0\vert$).

**Result (Zeno's paradox)**: As $\Delta t \to 0$ (infinite measurement rate), the system **remains** in $\vert\psi_0\rangle$; evolution is **frozen**.

**Mechanism**: Each measurement collapses the state back to the measured outcome. If measurements are frequent enough, evolution is prevented before the system can deviate significantly.

Consider the effect quantitatively:

:::{admonition} Derivation: Quantum Zeno Effect
:class: dropdown information

Between consecutive measurements the state evolves unitarily under $\hat{H}$. Expand the evolution operator to **second order** in the short interval $\Delta t$ — a first-order expansion is not enough, since it returns a survival probability greater than $1$:

$$
\vert\psi(\Delta t)\rangle = \mathrm{e}^{-\mathrm{i}\hat{H}\Delta t/\hbar}\vert\psi_0\rangle = \vert\psi_0\rangle - \frac{\mathrm{i}\Delta t}{\hbar}\hat{H}\vert\psi_0\rangle - \frac{(\Delta t)^2}{2\hbar^2}\hat{H}^2\vert\psi_0\rangle + O((\Delta t)^3).
$$

Projecting onto the initial state gives the survival amplitude

$$
\langle\psi_0\vert\psi(\Delta t)\rangle = 1 - \frac{\mathrm{i}\Delta t}{\hbar}\langle\hat{H}\rangle - \frac{(\Delta t)^2}{2\hbar^2}\langle\hat{H}^2\rangle + O((\Delta t)^3),
$$

where $\langle\hat{H}\rangle = \langle\psi_0\vert\hat{H}\vert\psi_0\rangle$ and $\langle\hat{H}^2\rangle = \langle\psi_0\vert\hat{H}^2\vert\psi_0\rangle$ are both real. The **survival probability** is the modulus squared of this amplitude. Writing it as $1 - \mathrm{i}a - b$ with real $a = \langle\hat{H}\rangle\Delta t/\hbar$ and $b = \langle\hat{H}^2\rangle(\Delta t)^2/(2\hbar^2)$,

$$
\begin{split}
\vert\langle\psi_0\vert\psi(\Delta t)\rangle\vert^2 &= (1-b)^2 + a^2 + O((\Delta t)^3) \\
&= 1 - 2b + a^2 + O((\Delta t)^3) \\
&= 1 - \frac{\langle\hat{H}^2\rangle}{\hbar^2}(\Delta t)^2 + \frac{\langle\hat{H}\rangle^2}{\hbar^2}(\Delta t)^2 + O((\Delta t)^3) \\
&= 1 - \frac{\langle\hat{H}^2\rangle - \langle\hat{H}\rangle^2}{\hbar^2}(\Delta t)^2 + O((\Delta t)^3),
\end{split}
$$

where the second line expands $(1-b)^2$ and drops $b^2$ as fourth order in $\Delta t$. The amplitude's linear term is purely imaginary ($-\mathrm{i}a$): it cancels in the modulus and contributes only through $a^2$ at second order, so the survival probability falls off **quadratically**, not linearly, in $\Delta t$.

The coefficient is the **energy variance** of $\hat{H}$ in the state $\vert\psi_0\rangle$:

$$
\begin{split}
\vert\langle\psi_0\vert\psi(\Delta t)\rangle\vert^2 &= 1 - c\,(\Delta t)^2, \\
c &= \frac{(\Delta H)^2}{\hbar^2}, \\
(\Delta H)^2 &= \langle\hat{H}^2\rangle - \langle\hat{H}\rangle^2.
\end{split}
$$

A projective measurement of $\hat{P} = \vert\psi_0\rangle\langle\psi_0\vert$ then finds the system still in $\vert\psi_0\rangle$ with probability $1 - c\,(\Delta t)^2$ and collapses it back to $\vert\psi_0\rangle$. The collapse resets the state, so the $n = T/\Delta t$ intervals over a total time $T$ are independent and identical:

$$
P(\text{remain in } \vert\psi_0\rangle) = \left(1 - c\,(\Delta t)^2\right)^{T/\Delta t}.
$$

Taking the logarithm and using $\ln(1 - x) \approx -x$ for small $x$,

$$
\ln P = \frac{T}{\Delta t}\ln\left(1 - c\,(\Delta t)^2\right) \approx -\frac{T}{\Delta t}\,c\,(\Delta t)^2 = -c\,T\,\Delta t,
$$

which vanishes as $\Delta t \to 0$. Hence $P \to 1$: in the limit of infinitely frequent measurement, evolution is **frozen**.
:::

:::{admonition} Discussion: The Quantum Zeno Effect
:class: dropdown tip

Frequent measurement freezes a state because the single-interval survival probability behaves as $1 - c(\Delta t)^2$—quadratic, not linear, in the measurement interval $\Delta t$. Why is this short-time quadratic behavior essential to the $\Delta t \to 0$ limit? If the system instead lost overlap linearly in $\Delta t$, as in pure exponential decay, what would happen to the survival probability over a fixed total time $T$? And in the opposite regime of rare measurements ($\Delta t \to \infty$), how does the picture reduce to ordinary unitary evolution? Discuss whether the resulting "freezing" reflects a physical action of measurement or simply repeated collapse.
:::

### Degenerate Eigenvalues

If eigenvalue $a_i$ is **degenerate** (multiple eigenstates), the projector projects onto the entire eigenspace:

$$
\hat{P}_i = \sum_{j=1}^{d_i} \vert u_j^{(i)}\rangle\langle u_j^{(i)}\vert
$$

where $d_i$ is the multiplicity (degeneracy).

**Post-measurement state**: 

$$
\hat{\rho}' = \frac{\hat{P}_i \hat{\rho} \hat{P}_i}{p_i}
$$

is supported on the entire eigenspace (possibly mixed, if initial state was a superposition of degenerate eigenstates).

:::{admonition} Example: Spin vs. Angular Momentum
:class: example

- **$\hat{Z}$ measurement**: Non-degenerate eigenvalues (+1, -1); outcome fully specifies the state.
- **$\hat{L}^2$ measurement** (orbital angular momentum squared): Degenerate eigenvalues ($l(l+1)\hbar^2$ with degeneracy $2l+1$); outcome specifies angular momentum but not magnetic quantum number.

:::

### Compatibility and Commuting Observables

Two observables $\hat{A}$ and $\hat{B}$ **commute** ($[\hat{A}, \hat{B}] = 0$) iff they have a **simultaneous eigenbasis**. In this case:
- Can measure both without disturbing the state (outcome of first doesn't affect second).
- Can assign definite values to both.

**Non-commuting observables** (e.g., $\hat{X}$ and $\hat{Z}$):
- Measuring $\hat{A}$ disturbs $\hat{B}$ (collapses to eigenstate of $\hat{A}$, which is superposition of $\hat{B}$ eigenstates).
- Cannot assign simultaneous definite values.
- Measurement outcomes are **incompatible**.

:::{admonition} Projective Measurement Is a Special Case
:class: note

Projective (von Neumann) measurements are the textbook case where outcomes correspond to orthogonal projectors summing to identity. The more general framework (POVM) relaxes orthogonality—see §6.3.2.
:::

### Summary

- **Spectral decomposition:** Every Hermitian observable $\hat{A} = \sum_i a_i \hat{P}_i$ decomposes into eigenvalues and orthogonal projectors onto eigenspaces.

- **Measurement postulate:** Outcome $a_i$ with probability $p_i = \operatorname{Tr}(\hat{P}_i \hat{\rho})$; state collapses to $\hat{P}_i \hat{\rho} \hat{P}_i / p_i$. For pure states, outcome probabilities follow the Born rule.

- **Repeated measurements:** Second measurement of the same observable gives the same outcome with certainty (reproducibility); no quantum evolution between measurements.

- **Quantum Zeno effect:** Frequent measurements suppress time evolution—as measurement interval $\Delta t \to 0$, the system freezes in its initial state.

- **Compatible observables:** If $[\hat{A}, \hat{B}] = 0$, they share an eigenbasis and can be measured simultaneously without disturbance. Non-commuting observables are mutually incompatible.
