## Lecture Notes

### Overview

Measurement in quantum mechanics is not a passive observation—it is a projective act that forces the system into one of several possible states. For any Hermitian observable, the measurement outcome is an eigenvalue, the probability depends on the initial state's alignment with the corresponding eigenspace (Born rule), and the post-measurement state is the normalized projection onto that eigenspace. This section develops the mathematical framework and explores key consequences: repeated measurements give the same outcome with certainty, frequent measurements can suppress evolution (quantum Zeno effect), and commuting observables can be measured simultaneously.

### Spectral Decomposition and Hermitian Operators

:::{admonition} Definition: Spectral Decomposition
:class: important

For a Hermitian operator $\hat{A}$ (observable), the **spectral theorem** states:

$$

A = \sum_i a_i P_i

$$ (eq-projmeas-spectral)

where:
- $a_i$ = eigenvalues (real, since $\hat{A}$ is Hermitian)
- $P_i$ = projector onto eigenspace of $a_i$
- $P_i$ are **orthogonal**: $P_i P_j = \delta_{ij} P_i$
- $P_i$ are **complete**: $\sum_i P_i = I$ (sum to identity)
- $P_i^2 = P_i$ (idempotent)

If eigenvalue $a_i$ is **degenerate** (multiplicity $> 1$), projector $P_i$ projects onto a multi-dimensional subspace.
:::

:::{admonition} Example: Qubit Spin Observable
:class: example

For $\hat{\sigma}^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$ (Pauli-z operator):

- Eigenvalues: $\lambda_1 = +1, \lambda_2 = -1$.
- Eigenvectors: $\vert 0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \vert 1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.
- Projectors:

$$

P_+ = \vert 0\rangle\langle 0\vert = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}, \quad P_- = \vert 1\rangle\langle 1\vert = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix}

$$

- Verification: $P_+ + P_- = I$ ✓, $P_+^2 = P_+$ ✓, $P_+ P_- = 0$ ✓.
- Spectral form: $\hat{\sigma}^z = (+1) P_+ + (-1) P_-$.

:::

### The Measurement Postulate

Recall the **measurement postulate** from §1.2.1: For a system in state $\hat{\rho}$ measured for observable $A = \sum_i a_i P_i$, the outcome is eigenvalue $a_i$ with probability $p_i = \text{Tr}(P_i \hat{\rho})$, and the state collapses to $\hat{\rho}' = P_i \hat{\rho} P_i / \text{Tr}(P_i \hat{\rho})$.

Here we explore the mathematical structure more carefully, generalizations, and physical consequences.

:::{admonition} Measurement Postulate: Probability and Collapse
:class: important

**Outcome**: Measurement yields eigenvalue $a_i$ with **probability**:

$$

p_i = \text{Tr}(P_i \hat{\rho})

$$ (eq-projmeas-probability)

**Post-measurement state** (state collapse):

$$

\hat{\rho}' = \frac{P_i \hat{\rho} P_i}{\text{Tr}(P_i \hat{\rho})} = \frac{P_i \hat{\rho} P_i}{p_i}

$$ (eq-projmeas-collapse)

The state is **projected** onto the eigenspace corresponding to the measured eigenvalue.
:::

#### Normalization Check

$$

\text{Tr}(\hat{\rho}') = \frac{\text{Tr}(P_i \hat{\rho} P_i)}{p_i} = \frac{p_i}{p_i} = 1 \quad ✓

$$

since $P_i^2 = P_i$ and $P_i$ is Hermitian.

:::{admonition} Derivation: Post-Measurement State Is a Valid Density Matrix
:class: dropdown information

We verify that the collapsed state $\hat{\rho}' = P_i \hat{\rho} P_i / p_i$ satisfies all three properties of a density matrix: unit trace, Hermiticity, and positive semi-definiteness.

**1. Trace = 1 (normalization).**

$$
\text{Tr}(\hat{\rho}') = \frac{\text{Tr}(P_i \hat{\rho} P_i)}{p_i}
$$

Using cyclicity of the trace and $P_i^2 = P_i$ (projector is idempotent):

$$
\text{Tr}(P_i \hat{\rho} P_i) = \text{Tr}(P_i^2 \hat{\rho}) = \text{Tr}(P_i \hat{\rho}) = p_i
$$

Therefore $\text{Tr}(\hat{\rho}') = p_i / p_i = 1$ $\checkmark$

**2. Hermiticity.**

Since $P_i$ is Hermitian ($P_i^\dagger = P_i$) and $\hat{\rho}$ is Hermitian ($\hat{\rho}^\dagger = \hat{\rho}$):

$$
(\hat{\rho}')^\dagger = \frac{(P_i \hat{\rho} P_i)^\dagger}{p_i} = \frac{P_i^\dagger \hat{\rho}^\dagger P_i^\dagger}{p_i} = \frac{P_i \hat{\rho} P_i}{p_i} = \hat{\rho}'
$$

So $\hat{\rho}'$ is Hermitian $\checkmark$

**3. Positive semi-definiteness.**

For any state $\vert\phi\rangle$:

$$
\langle\phi\vert \hat{\rho}' \vert\phi\rangle = \frac{\langle\phi\vert P_i \hat{\rho} P_i \vert\phi\rangle}{p_i} = \frac{\langle\psi\vert \hat{\rho} \vert\psi\rangle}{p_i}
$$

where $\vert\psi\rangle = P_i\vert\phi\rangle$. Since $\hat{\rho} \geq 0$, the numerator $\langle\psi\vert\hat{\rho}\vert\psi\rangle \geq 0$, and $p_i > 0$ (we conditioned on outcome $i$ occurring). Therefore $\hat{\rho}' \geq 0$ $\checkmark$

**Physical consequence.** The measurement postulate is self-consistent: repeated measurement of outcome $i$ with certainty is guaranteed since $\hat{\rho}'$ is supported on the eigenspace of $P_i$, giving $\text{Tr}(P_i \hat{\rho}') = \text{Tr}(P_i^2 \hat{\rho} P_i)/p_i = p_i/p_i = 1$.
:::

:::{admonition} Born Rule: Expectation Value
:class: important

The average outcome of repeated measurements is:

$$

\langle A \rangle = \sum_i a_i p_i = \sum_i a_i \text{Tr}(P_i \hat{\rho}) = \text{Tr}(A \hat{\rho})

$$ (eq-projmeas-expectation)

This is the **Born rule** in density matrix form.
:::

:::{admonition} Measurement of Pure States
:class: important

For a **pure state** $\hat{\rho} = \vert\psi\rangle\langle\psi\vert$:

**Probability**:

$$

p_i = \text{Tr}(P_i \vert\psi\rangle\langle\psi\vert) = \langle\psi\vert P_i \vert\psi\rangle

$$ (eq-projmeas-pure-prob)

If $\vert\psi\rangle = \sum_j c_j \vert u_j^{(i)}\rangle$ where $\vert u_j^{(i)}\rangle$ are eigenstates of eigenvalue $a_i$ (with multiplicity $d_i$):

$$

p_i = \sum_j \vert c_j\vert^2

$$

**Post-measurement state**:

$$

\vert\psi'\rangle = \frac{P_i \vert\psi\rangle}{\sqrt{p_i}}

$$ (eq-projmeas-pure-collapse)

The state is **normalized projection** onto eigenspace $i$.
:::

:::{admonition} Example: Measuring $\hat{\sigma}^z$
:class: example

Initial state: $\vert\psi\rangle = \frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$ (superposition).

Measure $\hat{\sigma}^z$:
- **Outcome** $+1$ ($\vert 0\rangle$):
  - Probability: $p_+ = \vert\frac{1}{\sqrt{2}}\vert^2 = 1/2$.
  - Post-measurement: $\vert\psi'\rangle = \vert 0\rangle$ (definite $\vert 0\rangle$).

- **Outcome** $-1$ ($\vert 1\rangle$):
  - Probability: $p_- = \vert\frac{1}{\sqrt{2}}\vert^2 = 1/2$.
  - Post-measurement: $\vert\psi'\rangle = \vert 1\rangle$ (definite $\vert 1\rangle$).

After measurement, the superposition is **destroyed**; the state is definite.

:::

### Repeated Measurements

**Theorem**: If you measure observable $A$ and obtain outcome $a_i$, then immediately measure $A$ again, you get the **same outcome** $a_i$ with **certainty** (probability 1).

**Proof**: After first measurement, state is $\hat{\rho}' = P_i \hat{\rho} P_i / p_i$, which is **supported** on eigenspace of $a_i$.

Second measurement probability:

$$

p_i^{(2)} = \text{Tr}(P_i \hat{\rho}') = \text{Tr}\left(P_i \frac{P_i \hat{\rho} P_i}{p_i}\right) = \frac{\text{Tr}(P_i P_i \hat{\rho} P_i)}{p_i} = \frac{\text{Tr}(P_i \hat{\rho} P_i)}{p_i} = \frac{p_i}{p_i} = 1

$$

**Consequence**: Measurement outcome is perfectly deterministic on repeated measurement; the result **doesn't change**.

:::{admonition} Discussion
:class: dropdown tip

The quantum Zeno effect says frequent measurements suppress transitions. For a two-level system with measurement interval $\tau$, compute the survival probability $P(T)$ as $\tau\to 0$ and $\tau\to\infty$.
:::

### The Quantum Zeno Effect

**Setup**: System starts in state $\vert\psi_0\rangle$ that would evolve to $\vert\psi(t)\rangle$ under unitary evolution (e.g., decay).

**Frequent projective measurements**: Measure at times $0, \Delta t, 2\Delta t, \ldots$ whether the system is still in $\vert\psi_0\rangle$ (projector $P = \vert\psi_0\rangle\langle\psi_0\vert$).

**Result (Zeno's paradox)**: As $\Delta t \to 0$ (infinite measurement rate), the system **remains** in $\vert\psi_0\rangle$; evolution is **frozen**.

**Mechanism**: Each measurement collapses the state back to the measured outcome. If measurements are frequent enough, evolution is prevented before the system can deviate significantly.

Consider the effect quantitatively:

:::{admonition} Derivation: Quantum Zeno Effect
:class: dropdown information

After time $\Delta t$ without measurement, state evolves to $\vert\psi(\Delta t)\rangle \approx \vert\psi_0\rangle - \mathrm{i} H \vert\psi_0\rangle \Delta t/\hbar$ (first order).

Overlap: $\vert\langle\psi_0\vert\psi(\Delta t)\rangle\vert^2 \approx 1 - O((\Delta t)^2)$.

After measurement: State either stays in $\vert\psi_0\rangle$ (probability $\approx 1 - O((\Delta t)^2)$) or exits (probability $\propto (\Delta t)^2$).

With $n = T/\Delta t$ measurements over time $T$:

$$

P(\text{remain in } \vert\psi_0\rangle) \approx (1 - c(\Delta t)^2)^{T/\Delta t} \to 1 \quad \text{as } \Delta t \to 0

$$

:::

### Degenerate Eigenvalues

If eigenvalue $a_i$ is **degenerate** (multiple eigenstates), the projector projects onto the entire eigenspace:

$$

P_i = \sum_{j=1}^{d_i} \vert u_j^{(i)}\rangle\langle u_j^{(i)}\vert

$$

where $d_i$ is the multiplicity (degeneracy).

**Post-measurement state**: 

$$

\hat{\rho}' = \frac{P_i \hat{\rho} P_i}{p_i}

$$

is supported on the entire eigenspace (possibly mixed, if initial state was a superposition of degenerate eigenstates).

:::{admonition} Example: Spin vs. Angular Momentum
:class: example

- **$\hat{\sigma}^z$ measurement**: Non-degenerate eigenvalues (+1, -1); outcome fully specifies the state.
- **$L^2$ measurement** (orbital angular momentum squared): Degenerate eigenvalues ($l(l+1)\hbar^2$ with degeneracy $2l+1$); outcome specifies angular momentum but not magnetic quantum number.

:::

### Compatibility and Commuting Observables

Two observables $A$ and $B$ **commute** ($[A, B] = 0$) iff they have a **simultaneous eigenbasis**. In this case:
- Can measure both without disturbing the state (outcome of first doesn't affect second).
- Can assign definite values to both.

**Non-commuting observables** (e.g., $\hat{\sigma}^x$ and $\hat{\sigma}^z$):
- Measuring $A$ disturbs $B$ (collapses to eigenstate of $A$, which is superposition of $B$ eigenstates).
- Cannot assign simultaneous definite values.
- Measurement outcomes are **incompatible**.


:::{admonition} Derivation: Commuting Operators Share a Simultaneous Eigenbasis
:class: dropdown information

**Claim:** $[\hat{A}, \hat{B}] = 0$ if and only if $\hat{A}$ and $\hat{B}$ have a common orthonormal eigenbasis.

**($\Leftarrow$) Shared eigenbasis $\Rightarrow$ commutativity.** If $\vert n\rangle$ satisfies $\hat{A}\vert n\rangle = a_n\vert n\rangle$ and $\hat{B}\vert n\rangle = b_n\vert n\rangle$, then on any such eigenstate $[\hat{A},\hat{B}]\vert n\rangle = (a_n b_n - b_n a_n)\vert n\rangle = 0$. Since the eigenstates span the space, $[\hat{A},\hat{B}] = 0$.

**($\Rightarrow$) Commutativity $\Rightarrow$ shared eigenbasis.** Let $\vert\psi\rangle$ be an eigenstate of $\hat{A}$ with eigenvalue $a$: $\hat{A}\vert\psi\rangle = a\vert\psi\rangle$. Apply $\hat{B}$:

$$
\hat{A}(\hat{B}\vert\psi\rangle) = \hat{B}(\hat{A}\vert\psi\rangle) = a(\hat{B}\vert\psi\rangle)
$$

So $\hat{B}\vert\psi\rangle$ is also an eigenstate of $\hat{A}$ with eigenvalue $a$: **$\hat{B}$ maps the eigenspace of $a$ into itself**.

Case 1 (non-degenerate $a$): The eigenspace is 1-dimensional, so $\hat{B}\vert\psi\rangle = b\vert\psi\rangle$ — $\vert\psi\rangle$ is automatically an eigenstate of $\hat{B}$ too.

Case 2 (degenerate $a$): $\hat{B}$ restricted to the eigenspace of $a$ is a Hermitian operator on that subspace. By the spectral theorem applied within the eigenspace, we can choose an orthonormal basis of common eigenstates of $\hat{A}$ and $\hat{B}$. Applying this to each eigenspace of $\hat{A}$ and assembling the results gives the full simultaneous eigenbasis. $\square$

**Physical consequence:** If $[\hat{A}, \hat{B}] = 0$, measuring $A$ leaves the system in an eigenstate of $B$ (or a superposition of degenerate $B$ eigenstates that are simultaneously $A$ eigenstates). The measurement outcomes are **compatible**: measuring $A$ then $B$ gives the same result as measuring $B$ then $A$.
:::

:::{admonition} Projective measurement is a special case
:class: note
Projective (von Neumann) measurements are the textbook case where outcomes correspond to orthogonal projectors summing to identity. The more general framework (POVM) relaxes orthogonality—see §6.3.2.
:::


:::{admonition} Derivation: Measurement Order Matters for Non-Commuting Observables
:class: dropdown information

**Claim.** For non-commuting observables $[A, B] \neq 0$, the joint probability $P(A{=}a_i \text{ then } B{=}b_j)$ generally differs from $P(B{=}b_j \text{ then } A{=}a_i)$.

**General formula.** From the measurement postulate applied sequentially:

$$
P(A{=}a_i\text{ then }B{=}b_j) = \mathrm{Tr}(P_j^B P_i^A \hat{\rho} P_i^A P_j^B)
= \mathrm{Tr}(P_j^B P_i^A \hat{\rho} P_i^A)
$$

using cyclicity of trace and $(P_j^B)^2 = P_j^B$. The reversed order gives:

$$
P(B{=}b_j\text{ then }A{=}a_i) = \mathrm{Tr}(P_i^A P_j^B \hat{\rho} P_j^B)
$$

These are equal iff $[P_i^A, P_j^B] = 0$ for all $i, j$, which holds iff $[A, B] = 0$.

**Concrete example: $\hat{\sigma}^x$ then $\hat{\sigma}^z$ vs. $\hat{\sigma}^z$ then $\hat{\sigma}^x$, starting from $\vert 0\rangle$.**

Since $[\hat{\sigma}^x, \hat{\sigma}^z] = -2\mathrm{i}\hat{\sigma}^y \neq 0$, we expect the two orders to differ.

**Order 1: Measure $\hat{\sigma}^x$ first, then $\hat{\sigma}^z$.**

Step 1 — $P(\sigma^x = +1)$: The eigenstate of $\sigma^x$ with eigenvalue $+1$ is $\vert+\rangle = \tfrac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle)$.

$$
P(\sigma^x{=}+1) = \vert\langle +\vert 0\rangle\vert^2 = \tfrac{1}{2}
$$

Post-measurement state: $\vert+\rangle$.

Step 2 — $P(\sigma^z = +1 \mid \text{state is }\vert+\rangle)$: The eigenstate of $\sigma^z$ with $+1$ is $\vert 0\rangle$.

$$
P(\sigma^z{=}+1\mid\sigma^x{=}+1) = \vert\langle 0\vert+\rangle\vert^2 = \tfrac{1}{2}
$$

Joint probability: $P(\sigma^x{=}+1\text{ then }\sigma^z{=}+1) = \tfrac{1}{2} \times \tfrac{1}{2} = \boxed{\tfrac{1}{4}}$.

**Order 2: Measure $\hat{\sigma}^z$ first, then $\hat{\sigma}^x$.**

Step 1 — $P(\sigma^z = +1)$: starting from $\vert 0\rangle$, $P(\sigma^z{=}+1) = 1$. Post-measurement state: $\vert 0\rangle$.

Step 2 — $P(\sigma^x = +1 \mid \text{state is }\vert 0\rangle)$: $P(\sigma^x{=}+1) = \vert\langle+\vert 0\rangle\vert^2 = \tfrac{1}{2}$.

Joint probability: $P(\sigma^z{=}+1\text{ then }\sigma^x{=}+1) = 1 \times \tfrac{1}{2} = \boxed{\tfrac{1}{2}}$.

**Conclusion.** $\tfrac{1}{4} \neq \tfrac{1}{2}$: the two measurement orders give different joint probabilities. This irreversibility of the measurement sequence is the quantum counterpart of non-commutativity.
:::


:::{admonition} Derivation: Complete Set of Commuting Observables (CSCO)
:class: dropdown information

A **complete set of commuting observables (CSCO)** is a maximal set of mutually commuting Hermitian operators $\{\hat{A}_1, \hat{A}_2, \ldots, \hat{A}_k\}$ whose simultaneous eigenvalues label every basis state uniquely. We prove that a CSCO always exists and uniquely determines the eigenbasis.

**Step 1: Mutually commuting operators share a joint eigenbasis.** If $[\hat{A}_i, \hat{A}_j] = 0$ for all $i, j$, then by the simultaneous diagonalizability theorem (extending "commuting operators share an eigenbasis" from §6.3.1), there exists an orthonormal basis $\{\vert\lambda\rangle\}$ in which every $\hat{A}_i$ is diagonal.

**Step 2: When is the labeling complete?** The simultaneous eigenvalues $(a_1, a_2, \ldots, a_k)$ uniquely label basis states if and only if no two basis states share the same tuple. Suppose $\hat{A}_1$ alone has a degenerate eigenspace (multiple states with eigenvalue $a_1$). Then we need $\hat{A}_2$ to split that degeneracy — i.e., the degenerate subspace of $\hat{A}_1$ must itself be non-degenerate under $\hat{A}_2$. This continues until the degeneracy is fully lifted.

**Step 3: Hydrogen atom example.** The Coulomb Hamiltonian $\hat{H}$ has degenerate energy levels $E_n$ with $n^2$-fold degeneracy. Adding $\hat{L}^2$ partially lifts the degeneracy (states labeled by $n, l$), and adding $\hat{L}_z$ labels them by $n, l, m$. Spin adds $s, m_s$. The set $\{\hat{H}, \hat{L}^2, \hat{L}_z, \hat{S}^2, \hat{S}_z\}$ is a CSCO for the hydrogen atom (without fine structure).

**Completeness criterion.** The CSCO is **complete** when the multiplicity of every joint eigenvalue $(a_1, \ldots, a_k)$ is exactly 1. The dimension of the simultaneous eigenspace determines how many more commuting operators are needed to complete the set.
:::

### Summary

- **Projective measurement**: von Neumann measurement based on spectral decomposition $A = \sum_i a_i P_i$.
- **Measurement postulate** (detailed in §1.2.1; extended here): Outcome $a_i$ with probability $p_i = \text{Tr}(P_i \hat{\rho})$; state collapses to $P_i \hat{\rho} P_i / p_i$.
- **Expectation value**: $\langle A \rangle = \text{Tr}(A\hat{\rho})$ (Born rule).
- **Pure states**: $p_i = \langle\psi\vert P_i \vert\psi\rangle$; post-measurement $\vert\psi'\rangle = P_i\vert\psi\rangle / \sqrt{p_i}$.
- **Repeated measurements**: Second measurement of same observable gives same outcome with certainty (measurement result is reproducible).
- **Quantum Zeno effect**: Frequent measurements freeze the state, preventing evolution.
- **Degeneracy**: Multiple eigenstates for same eigenvalue; projector onto entire eigenspace.
- **Commuting observables**: Share eigenbasis; can be measured simultaneously without disturbance.
- **Non-commuting**: Mutually incompatible; measuring one disturbs the other.
