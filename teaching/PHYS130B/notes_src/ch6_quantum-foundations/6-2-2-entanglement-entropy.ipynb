## Lecture Notes

### Overview

In [§6.2.1](6-2-1-product-and-entangled-states) we defined entanglement via the Schmidt rank: a bipartite pure state is entangled when its coefficient matrix has rank $\geq 2$. But rank alone cannot distinguish a barely-entangled state from a maximally-entangled one. To quantify *how much* entanglement a state carries, we need the **reduced density matrix** (obtained by the partial trace) and the **von Neumann entropy** (introduced in [§6.1.2](6-1-2-entropy)). Combining these gives the **entanglement entropy** — the central measure for pure bipartite states. The section also introduces Rényi entanglement entropies (generalizations that probe the full Schmidt spectrum) and the four Bell states (the maximally entangled two-qubit basis used throughout Chapter 6).

### Partial Trace and Reduced Density Matrix

:::{admonition} Definition: Partial Trace
:class: important

The **partial trace** over subsystem $B$ is the linear map $\mathrm{Tr}_B : \mathrm{End}(\mathcal{H}_{AB}) \to \mathrm{End}(\mathcal{H}_A)$ defined on product operators by:

$$
\mathrm{Tr}_B(\hat{A} \otimes \hat{B}) = \mathrm{Tr}(\hat{B})\,\hat{A}
$$

Tracing over $A$ is defined analogously, $\mathrm{Tr}_A(\hat{A} \otimes \hat{B}) = \mathrm{Tr}(\hat{A})\,\hat{B}$, and extends by linearity to all operators on $\mathcal{H}_{AB}$.

:::

:::{admonition} Example: Partial Trace in the Two-Qubit $4\times 4$ Picture
:class: dropdown example

Work in the computational product basis $\vert ij\rangle$ with $i,j\in\{0,1\}$ for qubits $A$ and $B$, ordered lexicographically $\vert 00\rangle,\vert 01\rangle,\vert 10\rangle,\vert 11\rangle$. Index rows and columns by $m,n\in\{0,1,2,3\}$ with $m=2i+j$ labeling $\vert ij\rangle$ (and the same for $n$). Then

$$
\hat{\rho}_{AB} =
\begin{pmatrix}
\rho_{00} & \rho_{01} & \rho_{02} & \rho_{03} \\
\rho_{10} & \rho_{11} & \rho_{12} & \rho_{13} \\
\rho_{20} & \rho_{21} & \rho_{22} & \rho_{23} \\
\rho_{30} & \rho_{31} & \rho_{32} & \rho_{33}
\end{pmatrix}.
$$

**Trace out $B$** ($\mathrm{Tr}_B$, leaving a $2\times 2$ matrix on $A$). With compound indices $m = 2i+j$ and $n = 2i'+j$,

$$
(\rho_A)_{ii'} = \sum_{j=0}^{1} \rho_{\,(2i+j),\,(2i'+j)}.
$$

Equivalently, view $\hat{\rho}_{AB}$ as a $2\times 2$ block matrix in the $A$ indices; then $(\rho_A)_{ii'}$ is the trace of the $(i,i')$ block (a $2\times 2$ matrix in $B$):

$$
\hat{\rho}_A =
\begin{pmatrix}
\rho_{00}+\rho_{11} & \rho_{02}+\rho_{13} \\
\rho_{20}+\rho_{31} & \rho_{22}+\rho_{33}
\end{pmatrix}.
$$

**Trace out $A$** ($\mathrm{Tr}_A$, leaving a $2\times 2$ matrix on $B$). With $m = 2i+j$ and $n = 2i+j'$,

$$
(\rho_B)_{jj'} = \sum_{i=0}^{1} \rho_{\,(2i+j),\,(2i+j')}.
$$

In the same $4\times 4$ labeling,

$$
\hat{\rho}_B =
\begin{pmatrix}
\rho_{00}+\rho_{22} & \rho_{01}+\rho_{23} \\
\rho_{10}+\rho_{32} & \rho_{11}+\rho_{33}
\end{pmatrix}.
$$

:::

:::{admonition} Definition: Reduced Density Matrix
:class: important

For a state $\hat{\rho}_{AB}$, the **reduced density matrix** on subsystem $A$ is

$$
\hat{\rho}_A = \mathrm{Tr}_B(\hat{\rho}_{AB})
$$ (eq-partial-trace)

**Physical meaning**: $\hat{\rho}_A$ encodes all predictions for measurements on subsystem $A$ alone:

$$
\langle \hat{O}_A \rangle = \mathrm{Tr}_A(\hat{\rho}_A \hat{O}_A) = \mathrm{Tr}_{AB}\bigl(\hat{\rho}_{AB}\, (\hat{O}_A \otimes \hat{I}_B)\bigr)
$$

:::

To compute $\hat{\rho}_A$ in practice, insert a complete basis $\{\vert j\rangle_B\}$ for subsystem $B$:

$$
\hat{\rho}_A = \sum_j \langle j\vert_B\, \hat{\rho}_{AB}\, \vert j\rangle_B
$$

:::{admonition} Example: Partial Trace of a Bell State
:class: dropdown example

**Problem.** Find the reduced density matrix $\hat{\rho}_A$ for the Bell state $\vert\Phi^+\rangle = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$.

**Solution.** The full density matrix is $\hat{\rho}_{AB} = \vert\Phi^+\rangle\langle\Phi^+\vert = \frac{1}{2}(\vert 00\rangle\langle 00\vert + \vert 00\rangle\langle 11\vert + \vert 11\rangle\langle 00\vert + \vert 11\rangle\langle 11\vert)$.

Trace over $B$ by inserting the complete $B$ basis $\{\vert 0\rangle_B, \vert 1\rangle_B\}$ as above. Each term factorizes as $\vert ij\rangle\langle kl\vert = (\vert i\rangle_A\langle k\vert)\otimes(\vert j\rangle_B\langle l\vert)$, so a $B$-sandwich acts only on the $B$ factor: $\langle m\vert_B\,(\vert ij\rangle\langle kl\vert)\,\vert m\rangle_B = \langle m\vert j\rangle\langle l\vert m\rangle\,\vert i\rangle_A\langle k\vert$. For $m = 0$,

$$
\begin{split}
\langle 0\vert_B\,\hat{\rho}_{AB}\,\vert 0\rangle_B &= \frac{1}{2}\bigl(\langle 0\vert 0\rangle\langle 0\vert 0\rangle\,\vert 0\rangle\langle 0\vert + \langle 0\vert 0\rangle\langle 1\vert 0\rangle\,\vert 0\rangle\langle 1\vert \\
&\quad + \langle 0\vert 1\rangle\langle 0\vert 0\rangle\,\vert 1\rangle\langle 0\vert + \langle 0\vert 1\rangle\langle 1\vert 0\rangle\,\vert 1\rangle\langle 1\vert\bigr) \\
&= \frac{1}{2}\bigl(\vert 0\rangle\langle 0\vert + 0 + 0 + 0\bigr) = \frac{1}{2}\vert 0\rangle\langle 0\vert
\end{split}
$$

Only the $\vert 00\rangle\langle 00\vert$ term survives. The two coherence terms $\vert 00\rangle\langle 11\vert$ and $\vert 11\rangle\langle 00\vert$ — the entanglement signature of $\vert\Phi^+\rangle$ — each carry one vanishing $B$-overlap ($\langle 1\vert 0\rangle = 0$ and $\langle 0\vert 1\rangle = 0$, respectively), and the $\vert 11\rangle\langle 11\vert$ term carries the vanishing factor $\langle 0\vert 1\rangle$ twice. The $m = 1$ sandwich works identically under $0\leftrightarrow 1$, leaving only the $\vert 11\rangle\langle 11\vert$ term:

$$
\langle 1\vert_B\,\hat{\rho}_{AB}\,\vert 1\rangle_B = \frac{1}{2}\vert 1\rangle\langle 1\vert
$$

Adding the two contributions,

$$
\hat{\rho}_A = \langle 0\vert_B\,\hat{\rho}_{AB}\,\vert 0\rangle_B + \langle 1\vert_B\,\hat{\rho}_{AB}\,\vert 1\rangle_B = \frac{1}{2}\vert 0\rangle\langle 0\vert + \frac{1}{2}\vert 1\rangle\langle 1\vert = \frac{\hat{I}}{2}
$$

The reduced state is maximally mixed, even though the total state $\vert\Phi^+\rangle$ is pure — a hallmark of maximal entanglement.
:::

**Connection to Schmidt Decomposition.** For a bipartite pure state in Schmidt form $\vert\Psi\rangle = \sum_k \lambda_k \vert u_k\rangle_A \vert v_k\rangle_B$ (see [§6.2.1](6-2-1-product-and-entangled-states)), the reduced density matrices are diagonal in the Schmidt bases:

$$
\begin{split}
\hat{\rho}_A &= \sum_k \lambda_k^2 \vert u_k\rangle_A\langle u_k\vert \\
\hat{\rho}_B &= \sum_k \lambda_k^2 \vert v_k\rangle_B\langle v_k\vert
\end{split}
$$

The eigenvalues $\lambda_k^2$ are the squared Schmidt coefficients — equal on both sides.

:::{admonition} Derivation: Reduced Density Matrix from the Schmidt Decomposition
:class: dropdown information

**Setup.** The Schmidt coefficients $\lambda_k$ are real and non-negative, so the bra reads $\langle\Psi\vert = \sum_l \lambda_l\,\langle u_l\vert_A\langle v_l\vert_B$. Forming the full density matrix and regrouping the $A$ and $B$ factors of each term,

$$
\hat{\rho}_{AB} = \vert\Psi\rangle\langle\Psi\vert = \sum_{k,l} \lambda_k\lambda_l\,\bigl(\vert u_k\rangle_A\langle u_l\vert\bigr)\otimes\bigl(\vert v_k\rangle_B\langle v_l\vert\bigr)
$$

**Trace out subsystem B.** Every term is a product operator, so the rule $\mathrm{Tr}_B(\hat{A}\otimes\hat{B}) = \mathrm{Tr}(\hat{B})\,\hat{A}$ from the Partial Trace definition above applies term by term:

$$
\hat{\rho}_A = \mathrm{Tr}_B(\hat{\rho}_{AB}) = \sum_{k,l} \lambda_k\lambda_l\,\vert u_k\rangle_A\langle u_l\vert\;\mathrm{Tr}\bigl(\vert v_k\rangle\langle v_l\vert\bigr)
$$

**Use Schmidt-basis orthonormality.** The Schmidt vectors $\{\vert v_k\rangle_B\}$ are orthonormal, so the leftover trace is $\mathrm{Tr}(\vert v_k\rangle\langle v_l\vert) = \langle v_l\vert v_k\rangle = \delta_{kl}$. The Kronecker delta collapses the double sum to its diagonal $k = l$ terms:

$$
\hat{\rho}_A = \sum_k \lambda_k^2\,\vert u_k\rangle_A\langle u_k\vert
$$

**The other subsystem.** Tracing out $A$ instead is the identical computation, now with $\mathrm{Tr}(\vert u_k\rangle\langle u_l\vert) = \langle u_l\vert u_k\rangle = \delta_{kl}$ from the orthonormality of $\{\vert u_k\rangle_A\}$:

$$
\hat{\rho}_B = \sum_k \lambda_k^2\,\vert v_k\rangle_B\langle v_k\vert
$$

Both reduced density matrices are already in spectral (diagonal) form, and they share the *same* eigenvalues — the squared Schmidt coefficients $\{\lambda_k^2\}$ — differing only in their eigenvectors ($\vert u_k\rangle_A$ for $A$, $\vert v_k\rangle_B$ for $B$). This equal spectrum is the key fact behind the entropy symmetry established in the next section.
:::

### Entanglement Entropy

The von Neumann entropy $S(\hat{\rho}) = -\mathrm{Tr}(\hat{\rho}\ln\hat{\rho})$ was introduced in [§6.1.2](6-1-2-entropy) as a measure of how mixed a quantum state is. Applied to the reduced density matrix, it becomes the **entanglement entropy**.

:::{admonition} Definition: Entanglement Entropy
:class: important

For a bipartite **pure** state $\vert\Psi\rangle_{AB}$ with Schmidt coefficients $\{\lambda_k\}$, the **entanglement entropy** is:

$$
S(A) = S(\hat{\rho}_A) = -\sum_k \lambda_k^2 \ln \lambda_k^2
$$ (eq-entanglement-entropy)

It quantifies the entanglement across the $A\vert B$ bipartition.
:::

**Key properties:**

- $S(A) = 0$ iff the state is a product (Schmidt rank 1; $\hat{\rho}_A$ is pure).
- $S(A) = \ln\min(d_A, d_B)$ for a maximally entangled state (all Schmidt coefficients equal).
- $S(A) = S(B)$ for any bipartite pure state (both reduced states share the same eigenvalue spectrum).
- $S$ is an **LOCC monotone**: it cannot increase under local operations and classical communication.

:::{admonition} Example: Partially Entangled State
:class: dropdown example

**Problem.** Find the entanglement entropy of $\vert\Psi(\theta)\rangle = \cos\theta\,\vert 00\rangle + \sin\theta\,\vert 11\rangle$.

**Solution.** Schmidt coefficients: $\lambda_1 = \cos\theta$, $\lambda_2 = \sin\theta$.

$$
S(\theta) = -\cos^2\theta\,\ln(\cos^2\theta) - \sin^2\theta\,\ln(\sin^2\theta)
$$

At $\theta = 0$: $S = 0$ (product). At $\theta = \pi/4$: $S = \ln 2$ (Bell state, maximally entangled).
:::

:::{admonition} Poll: Entanglement entropy for pure states
:class: dropdown poll

For a pure bipartite state $\vert\psi\rangle_{AB}$, the entanglement entropy is $S = -\mathrm{Tr}(\hat{\rho}_A \ln \hat{\rho}_A)$ where $\hat{\rho}_A = \mathrm{Tr}_B(\vert\psi\rangle\langle\psi\vert)$. When is $S = 0$?

(A) Only when $\vert\psi\rangle = \vert a\rangle_A \otimes \vert b\rangle_B$ (product state).

(B) When the reduced state $\hat{\rho}_A$ is pure but $\hat{\rho}_B$ is mixed.

(C) When the total state has zero entropy.

(D) Never — all entangled states have $S > 0$.

:::

### Rényi Entanglement Entropies

Entanglement entropy captures the *average* information lost when tracing out a subsystem. The **Rényi entropies** generalize this by probing the full spectrum of Schmidt coefficients at different "resolutions."

$$
S_\alpha(A) = \frac{1}{1-\alpha} \ln \sum_k \lambda_k^{2\alpha}
$$ (eq-renyi)

Special cases:

- $\alpha \to 1$: recovers the von Neumann entanglement entropy.
- $\alpha = 2$: $S_2 = -\ln\mathrm{Tr}(\hat{\rho}_A^2)$, experimentally accessible via the "swap trick" (measuring the overlap of two copies of $\hat{\rho}_A$).
- $\alpha = 0$: $S_0 = \ln r$ where $r$ is the Schmidt rank.
- $\alpha \to \infty$: $S_\infty = -\ln\lambda_{\max}^2$, dominated by the largest Schmidt coefficient.

### Bell States

:::{admonition} Definition: Bell States
:class: important

The four **Bell states** form an orthonormal basis for $\mathbb{C}^2 \otimes \mathbb{C}^2$ consisting entirely of maximally entangled states:

$$
\begin{split}
\vert\Phi^\pm\rangle &= \frac{1}{\sqrt{2}}(\vert 00\rangle \pm \vert 11\rangle) \\
\vert\Psi^\pm\rangle &= \frac{1}{\sqrt{2}}(\vert 01\rangle \pm \vert 10\rangle)
\end{split}
$$

Each has Schmidt rank 2 with $\lambda_1 = \lambda_2 = 1/\sqrt{2}$, giving $S = \ln 2$ and $\hat{\rho}_A = \hat{I}/2$.
:::

:::{admonition} Discussion: Mixed Reduced State Does Not Always Mean Entanglement
:class: dropdown tip

Bell states are maximally entangled, yet their reduced density matrices are completely mixed. Does a mixed reduced state always signal entanglement? Can a separable *mixed* state $\hat{\rho}_{AB}$ have a mixed reduced density matrix on one subsystem? (Hint: consider a product of mixed states, $\hat{\rho}_A \otimes \hat{\rho}_B$.)
:::

### Summary

- **Partial trace**: $\hat{\rho}_A = \mathrm{Tr}_B(\hat{\rho}_{AB})$ extracts subsystem information; eigenvalues are the squared Schmidt coefficients for pure bipartite states.

- **Entanglement entropy**: $S(A) = -\sum_k \lambda_k^2 \ln\lambda_k^2$ — the von Neumann entropy of the reduced state; zero iff product, maximum ($\ln r$) iff maximally entangled.

- **Entropy symmetry**: $S(A) = S(B)$ for any bipartite pure state — both reduced states share the same eigenvalue spectrum — a necessary consistency condition for $S$ to be a good entanglement measure of the $A\vert B$ cut.

- **Rényi entropies**: $S_\alpha$ generalize the von Neumann entropy and probe the full Schmidt spectrum; $S_2$ is experimentally accessible via the swap trick.

- **Bell states**: Four maximally entangled two-qubit states forming an orthonormal basis; $S = \ln 2$ each.
