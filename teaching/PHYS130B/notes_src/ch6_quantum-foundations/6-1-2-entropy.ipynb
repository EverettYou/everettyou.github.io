## Lecture Notes

### Overview

Von Neumann entropy $S(\hat{\rho}) = -\mathrm{Tr}(\hat{\rho} \ln \hat{\rho})$ quantifies how much we don't know about a quantum state. It is zero for pure states (perfect knowledge) and positive for mixed states (ignorance). The maximum entropy principle shows that thermal states—those that maximize entropy subject to a constraint on average energy—are precisely the Boltzmann distributions. This simple principle bridges quantum mechanics to thermodynamics without assuming temperature a priori.

### Von Neumann Entropy

:::{admonition} Von Neumann Entropy
:class: important

For a density matrix $\hat{\rho}$ on a $d$-dimensional Hilbert space, the **von Neumann entropy** is:

$$
S(\hat{\rho}) = -\mathrm{Tr}(\hat{\rho} \ln \hat{\rho})
$$ (eq-entropy-definition)

Equivalently, using spectral decomposition $\hat{\rho} = \sum_i \lambda_i |\psi_i\rangle\langle\psi_i|$ (eigenvalues $\lambda_i \geq 0$):

$$
S(\hat{\rho}) = -\sum_i \lambda_i \ln \lambda_i
$$ (eq-entropy-spectral)

with the convention $0 \ln 0 = 0$.
:::

**Bounds and limits:**

$$
0 \leq S(\hat{\rho}) \leq \ln d
$$ (eq-entropy-bounds)

The minimum $S = 0$ occurs for pure states (a single eigenvalue equals 1). The maximum $S = \ln d$ occurs for the maximally mixed state $\hat{\rho} = \hat{I}/d$ (all eigenvalues equal $1/d$).

:::{admonition} Example: Pure and Mixed States
:class: example

**Pure state:** $\hat{\rho} = |\psi\rangle\langle\psi|$ has one eigenvalue $\lambda_1 = 1$. Thus:

$$
S(\hat{\rho}) = -1 \cdot \ln 1 = 0
$$

**Maximally mixed qubit:** $\hat{\rho} = \hat{I}/2$ (2-dimensional) has eigenvalues $\lambda_1 = \lambda_2 = 1/2$. Thus:

$$
S(\hat{\rho}) = -2 \cdot \frac{1}{2} \ln\frac{1}{2} = \ln 2
$$

**Thermal qubit at finite temperature:** $\hat{\rho} = \frac{\mathrm{e}^{-\beta E_1} |1\rangle\langle 1| + \mathrm{e}^{-\beta E_2} |2\rangle\langle 2|}{Z}$ where $Z = \mathrm{e}^{-\beta E_1} + \mathrm{e}^{-\beta E_2}$. The entropy ranges between 0 (zero temperature, ground state only) and $\ln 2$ (infinite temperature, maximally mixed).
:::

:::{admonition} Discussion
:class: dropdown tip

A thermometer measures entropy by heat capacity. At high temperatures, a system has many accessible energy levels and entropy is large. At low temperatures, only the ground state is occupied and entropy vanishes. Why is entropy the right quantity to measure this ignorance?
:::


:::{admonition} Derivation: Entropy Bounds $0 \leq S \leq \ln d$
:class: dropdown information

**Lower bound $S \geq 0$:** For any eigenvalue $\lambda_i \in [0, 1]$, we have $\ln\lambda_i \leq 0$. Since $\lambda_i \geq 0$, each term $-\lambda_i\ln\lambda_i \geq 0$ (using $\lim_{\lambda\to 0^+} -\lambda\ln\lambda = 0$). So $S = -\sum_i \lambda_i\ln\lambda_i \geq 0$.

Equality ($S = 0$) requires every term to vanish, which happens when each $\lambda_i = 0$ or $\lambda_i = 1$. Combined with $\sum_i \lambda_i = 1$, exactly one $\lambda_i = 1$ and the rest are zero: a pure state.

**Upper bound $S \leq \ln d$:** We use the inequality $\ln x \leq x - 1$ (equality iff $x = 1$). Setting $x = (1/d)/\lambda_i$:

$$
\ln\frac{1}{d\lambda_i} \leq \frac{1}{d\lambda_i} - 1
$$

Multiply by $-\lambda_i$ and sum over $i$:

$$
S = -\sum_i \lambda_i\ln\lambda_i = \sum_i \lambda_i\ln\frac{1}{d\lambda_i} + \ln d \leq \sum_i \lambda_i\left(\frac{1}{d\lambda_i} - 1\right) + \ln d = \frac{n}{d} - 1 + \ln d
$$

Wait — more directly: apply the **Klein inequality** (concavity of $\ln$). By Jensen's inequality applied to the concave function $-x\ln x$:

$$
S = \sum_i (-\lambda_i\ln\lambda_i) \leq d\cdot \left(-\frac{1}{d}\ln\frac{1}{d}\right) = \ln d
$$

with equality when all $\lambda_i = 1/d$ (maximally mixed state $\hat{\rho} = \hat{I}/d$).
:::

### Properties of Entropy

:::{admonition} Properties of Von Neumann Entropy
:class: note

1. **Non-negativity:** $S(\hat{\rho}) \geq 0$, with equality iff $\hat{\rho}$ is a pure state.

2. **Upper bound:** $S(\hat{\rho}) \leq \ln d$ for $d$-dimensional systems, with equality iff $\hat{\rho} = \hat{I}/d$.

3. **Unitary invariance:** $S(U\hat{\rho} U^\dagger) = S(\hat{\rho})$ for any unitary $U$. Entropy is unchanged by reversible quantum operations.

4. **Concavity:** For $p \in [0,1]$,

$$
S(p\hat{\rho}_1 + (1-p)\hat{\rho}_2) \geq p S(\hat{\rho}_1) + (1-p) S(\hat{\rho}_2)
$$ (eq-entropy-concavity)

Mixing states increases entropy: ignorance about which mixture we're in adds to internal entropy.
:::

:::{admonition} Derivation: Concavity of Von Neumann Entropy
:class: dropdown information

We prove $S(p\hat{\rho}_1 + (1-p)\hat{\rho}_2) \geq pS(\hat{\rho}_1) + (1-p)S(\hat{\rho}_2)$ for any $p \in [0,1]$.

**Setup.** Introduce a classical ancilla qubit $A$ that records which state was prepared, forming the joint state on system $S$ and ancilla $A$:

$$
\hat{\rho}_{SA} = p\,\hat{\rho}_1^S \otimes \vert 0\rangle\langle 0\vert^A + (1-p)\,\hat{\rho}_2^S \otimes \vert 1\rangle\langle 1\vert^A
$$

**Step 1: Marginals.** Tracing out $A$ gives $\hat{\rho}_S = p\hat{\rho}_1 + (1-p)\hat{\rho}_2$. Tracing out $S$ gives $\hat{\rho}_A = p\vert 0\rangle\langle 0\vert + (1-p)\vert 1\rangle\langle 1\vert$, with entropy $S(\hat{\rho}_A) = H(p) = -p\ln p - (1-p)\ln(1-p)$ (binary Shannon entropy).

**Step 2: Entropy of $\hat{\rho}_{SA}$.** Since $\vert 0\rangle^A$ and $\vert 1\rangle^A$ are orthogonal, $\hat{\rho}_{SA}$ is block-diagonal with orthogonal blocks. Its eigenvalues are $\{p\lambda_i^{(1)}\}$ (from the $\hat{\rho}_1$ block) and $\{(1-p)\mu_j^{(2)}\}$ (from the $\hat{\rho}_2$ block). Computing the entropy:

$$
S(\hat{\rho}_{SA}) = -\sum_i p\lambda_i^{(1)}\ln(p\lambda_i^{(1)}) - \sum_j (1-p)\mu_j^{(2)}\ln((1-p)\mu_j^{(2)})
$$

Using $-p\lambda_i\ln(p\lambda_i) = -p\lambda_i\ln p - p\lambda_i\ln\lambda_i$ and $\sum_i \lambda_i = \sum_j \mu_j = 1$:

$$
S(\hat{\rho}_{SA}) = H(p) + p\,S(\hat{\rho}_1) + (1-p)\,S(\hat{\rho}_2)
$$

**Step 3: Subadditivity.** For any bipartite state, the von Neumann entropy satisfies **subadditivity**: $S(\hat{\rho}_{SA}) \leq S(\hat{\rho}_S) + S(\hat{\rho}_A)$. (This follows from the non-negativity of quantum relative entropy: $D(\hat{\rho}_{SA}\|\hat{\rho}_S \otimes \hat{\rho}_A) \geq 0$.)

**Step 4: Conclusion.** Combining Steps 2 and 3:

$$
H(p) + p\,S(\hat{\rho}_1) + (1-p)\,S(\hat{\rho}_2) \leq S(p\hat{\rho}_1 + (1-p)\hat{\rho}_2) + H(p)
$$

Canceling $H(p)$: $\quad p\,S(\hat{\rho}_1) + (1-p)\,S(\hat{\rho}_2) \leq S(p\hat{\rho}_1 + (1-p)\hat{\rho}_2)$. $\square$

**Physical meaning.** Mixing two states always increases (or preserves) entropy. This reflects our additional ignorance: we know neither which state the system is in, nor its internal uncertainty.
:::

### Connection to Information Theory

Von Neumann entropy is the quantum generalization of **Shannon entropy**. For a classical probability distribution $P = \{p_i\}$:

$$
H(P) = -\sum_i p_i \ln p_i
$$

If we measure $\hat{\rho}$ in basis $\{|\psi_i\rangle\}$ with outcomes $i$ having probability $p_i = \langle\psi_i|\hat{\rho}|\psi_i\rangle$, we obtain the Shannon entropy $H(P)$ of the measurement outcome distribution. But von Neumann entropy $S(\hat{\rho})$ is basis-independent and captures the intrinsic quantum uncertainty.

**Interpretation:** $S(\hat{\rho})$ measures how much we don't know about the state $\hat{\rho}$. 
- $S = 0$: We know the state exactly (pure state).
- $S > 0$: We have incomplete information; measuring may yield different outcomes.

:::{admonition} Derivation: Non-Negativity of Quantum Relative Entropy
:class: dropdown information

The **quantum relative entropy** between density matrices $\hat{\rho}$ and $\hat{\sigma}$ is:

$$

S(\hat{\rho}\|\hat{\sigma}) = \mathrm{Tr}(\hat{\rho}\ln\hat{\rho}) - \mathrm{Tr}(\hat{\rho}\ln\hat{\sigma})

$$

We prove $S(\hat{\rho}\|\hat{\sigma}) \geq 0$ with equality iff $\hat{\rho} = \hat{\sigma}$.

**Step 1: Spectral decompositions.** Write $\hat{\rho} = \sum_i p_i\vert i\rangle\langle i\vert$ and $\hat{\sigma} = \sum_j q_j\vert j\rangle\langle j\vert$ in their respective eigenbases. The **overlap matrix** $t_{ij} = \vert\langle i\vert j\rangle\vert^2$ is doubly stochastic: $t_{ij} \geq 0$, $\sum_j t_{ij} = 1$ for each $i$ (from completeness of $\{\vert j\rangle\}$).

**Step 2: Expand the relative entropy.** Expressing $\ln\hat{\sigma}$ in its eigenbasis:

$$

S(\hat{\rho}\|\hat{\sigma}) = \sum_i p_i \ln p_i - \sum_{i,j} p_i\,t_{ij}\ln q_j

$$

**Step 3: Apply Jensen's inequality.** Since $\ln$ is concave, Jensen's inequality gives $\sum_j t_{ij}\ln q_j \leq \ln\!\left(\sum_j t_{ij}q_j\right)$ for each fixed $i$. Therefore:

$$

S(\hat{\rho}\|\hat{\sigma}) \geq \sum_i p_i\ln p_i - \sum_i p_i\ln\!\left(\sum_j t_{ij}q_j\right) = \sum_i p_i \ln\frac{p_i}{\tilde{q}_i}

$$

where $\tilde{q}_i = \sum_j t_{ij}q_j$ defines a valid probability distribution (non-negative, $\sum_i \tilde{q}_i = \sum_j q_j = 1$).

**Step 4: Classical KL divergence is non-negative.** Using $\ln t \leq t - 1$ (equivalently $\mathrm{e}^x \geq 1 + x$) with $t = \tilde{q}_i/p_i$:

$$

\sum_i p_i \ln\frac{p_i}{\tilde{q}_i} = -\sum_i p_i\ln\frac{\tilde{q}_i}{p_i} \geq -\sum_i p_i\!\left(\frac{\tilde{q}_i}{p_i} - 1\right) = 1 - \sum_i \tilde{q}_i = 0

$$

Combining: $S(\hat{\rho}\|\hat{\sigma}) \geq 0$. $\square$

**Equality condition:** Both Jensen and KL steps saturate iff $t_{ij}$ is concentrated on single $j$ values and $\tilde{q}_i = p_i$ for all $i$ — forcing $\hat{\rho} = \hat{\sigma}$.

**Why this matters:** Subadditivity $S(\hat{\rho}_{SA}) \leq S(\hat{\rho}_S) + S(\hat{\rho}_A)$, used in the concavity proof above, follows directly by applying this result to $S(\hat{\rho}_{SA}\|\hat{\rho}_S\otimes\hat{\rho}_A) \geq 0$. Non-negativity of relative entropy is thus the engine behind most entropy inequalities in quantum information.
:::

### Maximum Entropy Principle

Among all density matrices with a fixed average energy $\langle E \rangle = \mathrm{Tr}(\hat{\rho} \hat{H})$, which one maximizes entropy?

:::{admonition} Derivation: Maximum Entropy and Thermal State
:class: dropdown information

We maximize $S(\hat{\rho}) = -\mathrm{Tr}(\hat{\rho} \ln \hat{\rho})$ subject to:
- $\mathrm{Tr}(\hat{\rho}) = 1$ (normalization)
- $\mathrm{Tr}(\hat{\rho} \hat{H}) = E$ (fixed average energy)

Using Lagrange multipliers with constraints, the extremum satisfies:

$$
\frac{\delta}{\delta \hat{\rho}}[S(\hat{\rho}) - \alpha \mathrm{Tr}(\hat{\rho}) - \beta \mathrm{Tr}(\hat{\rho} \hat{H})] = 0
$$

This gives $-\ln \hat{\rho} - \alpha \hat{I} - \beta \hat{H} = 0$, or:

$$
\hat{\rho} = \mathrm{e}^{-(\alpha + \beta \hat{H})}
$$

Applying normalization: $\mathrm{Tr}(\hat{\rho}) = 1$ defines the **partition function** $Z = \mathrm{Tr}(\mathrm{e}^{-\beta \hat{H}})$, so:

$$
\hat{\rho}_{\text{thermal}} = \frac{\mathrm{e}^{-\beta \hat{H}}}{Z}
$$

where $\beta = 1/(k_B T)$ is the **inverse temperature**.

The Lagrange multiplier $\beta$ is not arbitrary: it adjusts so that $\langle E \rangle = \mathrm{Tr}(\hat{\rho}_{\text{thermal}} \hat{H})$ matches the constraint.
:::

:::{admonition} Thermal State and Boltzmann Distribution
:class: important

The state maximizing entropy at fixed $\langle E \rangle$ is the **thermal state**:

$$
\hat{\rho}_{\text{thermal}} = \frac{\mathrm{e}^{-\beta \hat{H}}}{Z}
$$ (eq-entropy-thermal)

where $Z = \mathrm{Tr}(\mathrm{e}^{-\beta \hat{H}})$ is the **partition function** and $\beta = 1/(k_B T)$.

In the energy eigenbasis, diagonal elements are **Boltzmann weights**:

$$
P_n = \frac{\mathrm{e}^{-\beta E_n}}{Z}
$$ (eq-entropy-boltzmann)

where $E_n$ are energy eigenvalues.

**Key insight:** The Boltzmann distribution is not an assumption—it follows from the principle of maximum entropy. When we know only the average energy, the state of maximum ignorance (maximum entropy) is thermal.
:::

### Partition Function and Thermodynamics

The partition function $Z(\beta) = \mathrm{Tr}(\mathrm{e}^{-\beta \hat{H}})$ encodes all thermodynamic properties of a system at inverse temperature $\beta$.

**Average Energy:**

$$
\langle E \rangle = \mathrm{Tr}(\hat{\rho}_{\text{thermal}} \hat{H}) = -\frac{\mathrm{d} \ln Z}{\mathrm{d} \beta}
$$ (eq-entropy-average-energy)

**Free Energy (Helmholtz):**

$$
F = -k_B T \ln Z
$$ (eq-entropy-free-energy)

where $T = 1/(k_B \beta)$.

**Entropy in terms of $Z$:**

From $\hat{\rho}_{\text{thermal}} = \mathrm{e}^{-\beta \hat{H}}/Z$:

$$
S(\hat{\rho}_{\text{thermal}}) = k_B [\beta \langle E \rangle + \ln Z]
$$ (eq-entropy-from-partition)

Equivalently, $S = -(\partial F/\partial T)|_V$ (standard thermodynamic relation).

**Thermodynamic Identities:**

The first law connects all three:

$$
\mathrm{d}F = -S \, \mathrm{d}T - P \, \mathrm{d}V
$$

(ignoring volume dependence for simplicity). From $F = \langle E \rangle - TS$:

$$
S = \frac{\langle E \rangle - F}{T} = \frac{\langle E \rangle + k_B T \ln Z}{T}
$$

which is consistent with equation {eq}`eq-entropy-from-partition`.


:::{admonition} Derivation: Thermodynamic Identities from the Partition Function
:class: dropdown information

Start from $Z(\beta) = \operatorname{Tr}(\mathrm{e}^{-\beta\hat{H}}) = \sum_n \mathrm{e}^{-\beta E_n}$ and $\hat{\rho} = \mathrm{e}^{-\beta\hat{H}}/Z$.

**Average energy.** Differentiate $\ln Z$:

$$
-\frac{\mathrm{d}\ln Z}{\mathrm{d}\beta} = -\frac{1}{Z}\frac{\mathrm{d}Z}{\mathrm{d}\beta} = -\frac{1}{Z}\operatorname{Tr}\!\left(-\hat{H}\,\mathrm{e}^{-\beta\hat{H}}\right) = \operatorname{Tr}(\hat{\rho}\hat{H}) = \langle E\rangle
$$

So $\langle E\rangle = -\partial(\ln Z)/\partial\beta$.

**Entropy.** Using $\ln\hat{\rho} = -\beta\hat{H} - (\ln Z)\hat{I}$:

$$
S = -\operatorname{Tr}(\hat{\rho}\ln\hat{\rho}) = -\operatorname{Tr}(\hat{\rho}(-\beta\hat{H} - \ln Z)) = \beta\langle E\rangle + \ln Z
$$

**Free energy.** Multiply by $k_B T = 1/\beta$:

$$
k_B T S = k_B T \beta\langle E\rangle + k_B T\ln Z \implies \langle E\rangle - TS = -k_B T\ln Z \equiv F
$$

The free energy $F = \langle E\rangle - TS$ is minimized at thermodynamic equilibrium — the thermal state achieves the minimum free energy (equivalently, maximum entropy) simultaneously.
:::

### Summary

- **Von Neumann entropy** $S(\hat{\rho}) = -\mathrm{Tr}(\hat{\rho} \ln \hat{\rho}) = -\sum_i \lambda_i \ln \lambda_i$ (where $\lambda_i$ are eigenvalues).
  - $S = 0$ for pure states; $S = \ln d$ for maximally mixed state in $d$ dimensions.
  - Ranges from 0 to $\ln d$; basis-independent; increased by mixing.

- **Maximum entropy principle:** Among all $\hat{\rho}$ with fixed $\langle E \rangle$, the one maximizing $S$ is the **thermal state** $\hat{\rho} = \mathrm{e}^{-\beta \hat{H}}/Z$.
  - This derives the **Boltzmann distribution** $P_n \propto \mathrm{e}^{-\beta E_n}$ from maximum ignorance, not as an assumption.

- **Partition function** $Z = \mathrm{Tr}(\mathrm{e}^{-\beta \hat{H}})$ bridges quantum mechanics to thermodynamics:
  - Free energy: $F = -k_B T \ln Z$
  - Average energy: $\langle E \rangle = -\mathrm{d}(\ln Z)/\mathrm{d}\beta$
  - Entropy: $S = k_B(\beta \langle E \rangle + \ln Z)$
