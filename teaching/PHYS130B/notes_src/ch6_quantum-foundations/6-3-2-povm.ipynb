## Lecture Notes

### Overview

A key limitation of projective measurements is that they cannot reliably distinguish non-orthogonal quantum states. If we relax the requirement that measurement operators be projectors and allow general positive operators—as long as they sum to the identity—we gain flexibility to improve state discrimination. This generalization defines **positive operator-valued measures (POVMs)**.

### POVM Definition

Where a projective measurement assigns each outcome a mutually orthogonal eigenspace, a POVM relaxes that orthogonality constraint, keeping only the conditions needed for a valid probability rule.

A **positive operator-valued measure (POVM)** is a set of operators $\{\hat{M}_1, \hat{M}_2, \ldots, \hat{M}_n\}$ on Hilbert space $\mathcal{H}$ satisfying two conditions:

**1. Positive semidefiniteness**:

$$
\hat{M}_i \geq 0 \quad \text{for all } i \quad (\text{all eigenvalues} \geq 0)
$$ (eq-povm-positive)

**2. Completeness**:

$$
\sum_i \hat{M}_i = \hat{I}
$$ (eq-povm-complete)

These conditions ensure that measurement probabilities are non-negative and sum to one.

### Measurement with a POVM

If a system in state $\hat{\rho}$ is measured using POVM $\{\hat{M}_i\}$:

**Outcome probability**:

$$
p_i = \operatorname{Tr}(\hat{M}_i \hat{\rho})
$$ (eq-povm-probability)

**Post-measurement state**:

$$
\hat{\rho}'_i = \frac{\sqrt{\hat{M}_i} \hat{\rho} \sqrt{\hat{M}_i}}{\operatorname{Tr}(\hat{M}_i \hat{\rho})}
$$ (eq-povm-collapse)

where $\sqrt{\hat{M}_i}$ is the unique positive square root (defined via spectral decomposition: $\hat{M}_i = \sum_j \lambda_{i,j} \vert v_{i,j}\rangle\langle v_{i,j}\vert  \Rightarrow \sqrt{\hat{M}_i} = \sum_j \sqrt{\lambda_{i,j}} \vert v_{i,j}\rangle\langle v_{i,j}\vert$).

This post-measurement-state formula is the **Lüders update**, the canonical state-update rule for a POVM. Unlike the outcome probability, however, the post-measurement state is **not** fixed by the POVM alone: the same element $\hat{M}_i$ can be implemented by different physical measurements that reproduce its outcome statistics yet leave the system in different states. The Lüders form is the standard choice taken when no further detail of the measuring apparatus is specified.

**Normalization check**: $\sum_i p_i = \sum_i \operatorname{Tr}(\hat{M}_i \hat{\rho}) = \operatorname{Tr}(\hat{I} \hat{\rho}) = 1$ ✓

### POVM vs. Projective Measurement

A **projective measurement** is a special case of a POVM where elements are orthogonal projectors $\hat{M}_i = \hat{P}_i$ with:
- $\hat{P}_i^2 = \hat{P}_i$ (idempotent)
- $\hat{P}_i \hat{P}_j = \delta_{ij} \hat{P}_i$ (orthogonal)
- $\sum_i \hat{P}_i = \hat{I}$ (complete)

**Key differences:**

| Property | Projective Measurement | General POVM |
|----------|----------------------|--------------|
| **Outcomes** | Limited to $d$ (Hilbert space dimension) | Can have $n > d$ (overcomplete) |
| **Orthogonality** | $\hat{M}_i \hat{M}_j = 0$ for $i \neq j$ | Elements can overlap: $\hat{M}_i \hat{M}_j \neq 0$ |
| **Information** | Binary outcome per qubit | More outcomes = more information per measurement |
| **Rank** | Rank-1 projectors | Can be rank-1 or higher |
| **Physical meaning of outcome** | Eigenvalue of measured observable | Index number (no direct physical meaning) |

**Information trade-off**: General POVMs extract more information about the system by allowing non-orthogonal measurement elements, at the cost of less definite outcome meaning.

### Unambiguous State Discrimination

One of the most striking applications of POVMs: **distinguishing non-orthogonal quantum states with certainty** when a definite answer is given.

**Problem**: Given a qubit in one of two non-orthogonal states $\vert \psi_1\rangle$ or $\vert \psi_2\rangle$ with overlap $\langle\psi_1\vert \psi_2\rangle = c$ (where $0 < \vert c\vert < 1$), can you determine which state with **certainty** (never make an error)?

**Projective measurement**: No. Non-orthogonal states have overlapping eigenspaces, so any projective measurement will sometimes give wrong answers.

**POVM solution**: Yes, with a trade-off. Use a **three-outcome POVM**:
- Outcome 1: "State is $\vert \psi_1\rangle$" (definite, no error)
- Outcome 2: "State is $\vert \psi_2\rangle$" (definite, no error)
- Outcome 3: "Inconclusive" (try again or use different strategy)

:::{admonition} Construction: Unambiguous State Discrimination POVM
:class: dropdown note

Work in the two-dimensional qubit space. Given two states $\vert\psi_1\rangle, \vert\psi_2\rangle$ with overlap $c = \langle\psi_1\vert\psi_2\rangle$, $0 < \vert c\vert < 1$, write $\vert\psi_1^\perp\rangle$ and $\vert\psi_2^\perp\rangle$ for the unit vectors orthogonal to $\vert\psi_1\rangle$ and $\vert\psi_2\rangle$.

**Step 1 — the unambiguous elements.** Define the first two elements with a normalization constant $a > 0$ left to be fixed:

$$
\begin{split}
\hat{M}_1 &= a\,(\hat{I} - \vert\psi_2\rangle\langle\psi_2\vert), \\
\hat{M}_2 &= a\,(\hat{I} - \vert\psi_1\rangle\langle\psi_1\vert).
\end{split}
$$

In two dimensions $\{\vert\psi_2\rangle, \vert\psi_2^\perp\rangle\}$ is an orthonormal basis, so its completeness relation $\vert\psi_2\rangle\langle\psi_2\vert + \vert\psi_2^\perp\rangle\langle\psi_2^\perp\vert = \hat{I}$ gives the identity

$$
\hat{I} - \vert\psi_2\rangle\langle\psi_2\vert = \vert\psi_2^\perp\rangle\langle\psi_2^\perp\vert,
$$

and likewise $\hat{I} - \vert\psi_1\rangle\langle\psi_1\vert = \vert\psi_1^\perp\rangle\langle\psi_1^\perp\vert$. Hence $\hat{M}_1 = a\,\vert\psi_2^\perp\rangle\langle\psi_2^\perp\vert$ and $\hat{M}_2 = a\,\vert\psi_1^\perp\rangle\langle\psi_1^\perp\vert$ are positive rank-one operators. They are designed so that

$$
\langle\psi_2\vert\hat{M}_1\vert\psi_2\rangle = a\,\vert\langle\psi_2^\perp\vert\psi_2\rangle\vert^2 = 0,
$$

and similarly $\langle\psi_1\vert\hat{M}_2\vert\psi_1\rangle = 0$: outcome 1 can never be triggered by $\vert\psi_2\rangle$, and outcome 2 never by $\vert\psi_1\rangle$. This is what makes the two identifications error-free.

**Step 2 — fix the normalization from positivity of $\hat{M}_3$.** The third element is the remainder,

$$
\hat{M}_3 = \hat{I} - \hat{M}_1 - \hat{M}_2,
$$

assigned to the inconclusive outcome. Completeness $\hat{M}_1 + \hat{M}_2 + \hat{M}_3 = \hat{I}$ then holds automatically; the binding constraint is positivity, $\hat{M}_3 \geq 0$, and this is what fixes $a$. Since $\hat{M}_3 = \hat{I} - (\hat{M}_1 + \hat{M}_2)$, it is positive semidefinite exactly when every eigenvalue of $\hat{M}_1 + \hat{M}_2$ is at most $1$.

Evaluate that spectrum in the orthonormal basis $\{\vert 0\rangle, \vert 1\rangle\}$ in which the two states take the symmetric form

$$
\begin{split}
\vert\psi_1\rangle &= \cos\tfrac{\theta}{2}\,\vert 0\rangle + \sin\tfrac{\theta}{2}\,\vert 1\rangle, \\
\vert\psi_2\rangle &= \cos\tfrac{\theta}{2}\,\vert 0\rangle - \sin\tfrac{\theta}{2}\,\vert 1\rangle,
\end{split}
$$

with $\langle\psi_1\vert\psi_2\rangle = \cos\theta = \vert c\vert$. (Any pair of states with overlap modulus $\vert c\vert$ can be brought to this form by a unitary together with phase redefinitions, none of which changes the eigenvalues of $\hat{M}_3$.) The orthogonal partners are $\vert\psi_2^\perp\rangle = \sin\tfrac{\theta}{2}\vert 0\rangle + \cos\tfrac{\theta}{2}\vert 1\rangle$ and $\vert\psi_1^\perp\rangle = \sin\tfrac{\theta}{2}\vert 0\rangle - \cos\tfrac{\theta}{2}\vert 1\rangle$. The two rank-one projectors carry off-diagonal entries $\pm a\sin\tfrac{\theta}{2}\cos\tfrac{\theta}{2}$ of opposite sign, so the cross terms cancel:

$$
\hat{M}_1 + \hat{M}_2 = a\begin{pmatrix} 2\sin^2\tfrac{\theta}{2} & 0 \\ 0 & 2\cos^2\tfrac{\theta}{2} \end{pmatrix} = a\begin{pmatrix} 1 - \vert c\vert & 0 \\ 0 & 1 + \vert c\vert \end{pmatrix}.
$$

Its largest eigenvalue is $a(1 + \vert c\vert)$, so $\hat{M}_3 \geq 0$ requires $a(1 + \vert c\vert) \leq 1$. The sharpest discrimination takes the largest admissible value,

$$
a = \frac{1}{1 + \vert c\vert},
$$

for which $\hat{M}_3$ has eigenvalues $1 - a(1 + \vert c\vert) = 0$ and $1 - a(1 - \vert c\vert) = \tfrac{2\vert c\vert}{1 + \vert c\vert}$ — positive semidefinite, with the smallest eigenvalue exactly zero. The completed POVM is

$$
\begin{split}
\hat{M}_1 &= \frac{\hat{I} - \vert\psi_2\rangle\langle\psi_2\vert}{1 + \vert c\vert}, \\
\hat{M}_2 &= \frac{\hat{I} - \vert\psi_1\rangle\langle\psi_1\vert}{1 + \vert c\vert}, \\
\hat{M}_3 &= \hat{I} - \hat{M}_1 - \hat{M}_2.
\end{split}
$$

**Step 3 — outcome probabilities.** Suppose the input is $\vert\psi_1\rangle$. The correct identification has probability

$$
\begin{split}
P(1\vert\psi_1) = \langle\psi_1\vert\hat{M}_1\vert\psi_1\rangle &= \frac{1}{1 + \vert c\vert}\,\langle\psi_1\vert(\hat{I} - \vert\psi_2\rangle\langle\psi_2\vert)\vert\psi_1\rangle \\
&= \frac{1 - \vert\langle\psi_1\vert\psi_2\rangle\vert^2}{1 + \vert c\vert} = \frac{1 - \vert c\vert^2}{1 + \vert c\vert} = 1 - \vert c\vert.
\end{split}
$$

The wrong identification never happens, $P(2\vert\psi_1) = \langle\psi_1\vert\hat{M}_2\vert\psi_1\rangle = 0$ (Step 1), so the inconclusive outcome carries the rest,

$$
P(3\vert\psi_1) = 1 - P(1\vert\psi_1) - P(2\vert\psi_1) = \vert c\vert.
$$

By the $\psi_1 \leftrightarrow \psi_2$ symmetry of the construction, an input $\vert\psi_2\rangle$ gives $P(2\vert\psi_2) = 1 - \vert c\vert$, $P(1\vert\psi_2) = 0$, and $P(3\vert\psi_2) = \vert c\vert$.

**Key property.** Each state is identified with **zero error** — outcomes 1 and 2 are never false — at the cost of an inconclusive outcome 3 occurring with probability $\vert c\vert$, exactly the overlap modulus. The more similar the states (larger $\vert c\vert$), the less often the discrimination succeeds; orthogonal states ($\vert c\vert \to 0$) are never inconclusive.

:::

### Naimark's Theorem: POVM as Extended Projective Measurement

Every POVM on a Hilbert space can be realized as a projective measurement on an **extended Hilbert space** (original space plus auxiliary system).

:::{admonition} Naimark's Theorem
:class: important

Every POVM $\{\hat{M}_i\}$ on $\mathcal{H}$ can be realized as a projective measurement on an extended space $\mathcal{H} \otimes \mathcal{H}_\text{aux}$. Specifically, there exist:
- An auxiliary (ancilla) Hilbert space $\mathcal{H}_\text{aux}$
- Orthogonal projectors $\{\hat{P}_i\}$ on $\mathcal{H} \otimes \mathcal{H}_\text{aux}$

such that:

$$
\hat{M}_i = \operatorname{Tr}_\text{aux}(\hat{P}_i (\hat{I}_{\mathcal{H}} \otimes \vert 0\rangle\langle 0\vert_\text{aux}))
$$ (eq-povm-neumark)

**Physical picture**: Entangle the system with an ancilla, perform a projective measurement on the combined system, and trace out (discard) the ancilla measurement outcome.

:::

This theorem reveals a profound truth: **POVMs are not fundamentally more general than projective measurements**—they are equivalent to projective measurements on a larger system. The apparent "softness" of a POVM comes from not observing (or not having access to) the ancilla measurement outcome.

### Examples

:::{admonition} Example: Six-State Octahedral IC-POVM on a Qubit
:class: dropdown example

One convenient informationally complete POVM for a qubit uses the six eigenstates of $\hat{\sigma}^{x}$, $\hat{\sigma}^{y}$, and $\hat{\sigma}^{z}$. These states sit at the vertices of a regular **octahedron** on the Bloch sphere. Define

$$
\hat{M}_{k} = \tfrac{1}{3}\vert\phi_{k}\rangle\langle\phi_{k}\vert\quad (k = 1,\ldots,6).
$$

This is a valid POVM because each Pauli axis contributes a pair of opposite projectors summing to $\hat{I}$:

$$
\sum_{k=1}^{6}\hat{M}_{k}
= \frac{1}{3}\sum_{i=x,y,z}\left(\vert{+}_{i}\rangle\langle{+}_{i}\vert + \vert{-}_{i}\rangle\langle{-}_{i}\vert\right)
= \hat{I}.
$$

The six probabilities are informationally complete: they contain enough data to reconstruct the Bloch vector of any qubit state. This octahedral POVM is therefore useful for tomography from three mutually unbiased bases.

A different standard construction is the qubit **SIC-POVM**, which uses four tetrahedral states (the minimal number $d^{2}=4$ for informational completeness in dimension $d=2$). Both are IC-POVMs; the octahedral version is overcomplete, while the SIC-POVM is minimal and symmetric.
:::

:::{admonition} Example: Heterodyne Detection (Optical POVM)
:class: dropdown example

In quantum optics, measuring quadrature amplitudes (amplitude and phase of a light field) is described by a continuous POVM:

$$
\hat{M}(\alpha) = \frac{1}{\pi}\vert \alpha\rangle\langle\alpha\vert_\text{coherent}
$$

where $\vert \alpha\rangle$ are coherent states indexed by complex amplitude $\alpha \in \mathbb{C}$.

**Properties**:
- Continuous outcome space (not discrete)
- Coherent states are highly non-orthogonal: $\vert\langle\alpha\vert \beta\rangle\vert = \mathrm{e}^{-\vert \alpha-\beta\vert^2/2} \neq 0$ for $\alpha \neq \beta$
- Enables **full quantum tomography** of optical field states

:::

### Summary

- **POVM definition:** Positive operator-valued measures $\{\hat{M}_m\}$ satisfy $\hat{M}_m \geq 0$ and $\sum_m \hat{M}_m = \hat{I}$, but are not required to be orthogonal or satisfy $\hat{M}_m^2 = \hat{M}_m$ (unlike projectors).

- **Measurement probabilities:** $p_m = \operatorname{Tr}(\hat{M}_m \hat{\rho})$ (same formula as projective), but post-measurement state is **not** determined by POVM alone—different physical realizations can give same outcome statistics.

- **Advantages over projectors:** More general; one outcome can have multiple post-measurement states (depends on probe details); allows finer-grained outcome statistics with fewer outcomes than dimension of Hilbert space.

- **Unambiguous state discrimination:** POVM enables non-projective measurements that sometimes identify unknown quantum states with zero ambiguity (no error-free projective equivalent exists).

- **Naimark's theorem:** Every POVM on $\mathcal{H}$ can be realized as an ordinary projective measurement on an extended space $\mathcal{H} \otimes \mathcal{H}_\text{aux}$—entangle the system with an ancilla, measure the enlarged system projectively, and discard the ancilla outcome. POVMs are therefore not fundamentally more general than projective measurements; their extra flexibility is the price of not observing the ancilla.
