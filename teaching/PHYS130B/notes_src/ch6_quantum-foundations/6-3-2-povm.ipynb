## Lecture Notes

Projective measurements are elegant but restrictive: outcomes must be orthogonal, and we learn something definite about the system. Yet in many practical scenarios, we wish to perform softer measurements that leave less disturbance, or distinguish between non-orthogonal states. Positive operator-valued measures generalize the measurement postulate, enabling a vast new toolkit for quantum information processing and state discrimination.

### Definition of POVM



:::{admonition} Discussion
:class: dropdown warning

A POVM with 3 elements can be implemented by a projective measurement on a larger system (Naimark dilation). For the trine POVM {M_1, M_2, M_3}, explicitly construct the dilation.
:::


A **positive operator-valued measure** (POVM) is a set of operators $\{M_1, M_2, \ldots, M_n\}$ on Hilbert space $\mathcal{H}$ satisfying:

**1. Positive semidefiniteness**:

$$


M_i \geq 0 \quad \text{for all } i \quad (\text{eigenvalues} \geq 0)

$$ (eq-povm-positive)

**2. Completeness**:

$$


\sum_i M_i = I

$$ (eq-povm-complete)

### Measurement Process with POVM

If a system in state $\rho$ is measured with POVM $\{M_i\}$:

**Outcome**: Measurement yields result $i$ with **probability**:

$$


p_i = \text{Tr}(M_i \rho)

$$ (eq-povm-probability)

**Post-measurement state** (state after measurement):

$$


\rho' = \frac{\sqrt{M_i} \rho \sqrt{M_i}}{\text{Tr}(M_i \rho)}

$$ (eq-povm-collapse)

The post-measurement state involves $\sqrt{M_i}$ (square root of $M_i$), not the projector itself.

#### Normalization

$$


\sum_i p_i = \sum_i \text{Tr}(M_i \rho) = \text{Tr}\left(\sum_i M_i \rho\right) = \text{Tr}(I \rho) = 1 \quad ✓

$$


### POVM vs. Projective Measurement

**Projective measurement** (special case of POVM):
- POVM elements are **orthogonal projectors**: $M_i = P_i$, with $P_i P_j = \delta_{ij} P_i$.
- Post-measurement state: $\rho' = P_i \rho P_i / p_i$ (same as before, since $P_i^2 = P_i$).
- Outcomes correspond to eigenvalues of Hermitian operator.

**General POVM**:
- Elements **need not be orthogonal**: $M_i M_j \neq 0$ in general.
- More outcomes possible than dimensions: $n > \dim(\mathcal{H})$ allowed (overcomplete basis).
- Can extract more information than orthogonal projections.

#### Trade-off

Projective measurement:
- **Pro**: Outcomes are eigenvalues (physical meaning); state collapse onto eigenstates.
- **Con**: Limited to $d$ outcomes for $d$-dimensional space; cannot measure non-orthogonal states unambiguously.

POVM:
- **Pro**: $n > d$ outcomes possible; unambiguous state discrimination; more information.
- **Con**: Outcomes are indices, not eigenvalues; post-measurement states more complex.

### Unambiguous State Discrimination

**Problem**: Given a qubit in one of two non-orthogonal states $|\psi_1\rangle$ or $|\psi_2\rangle$, can you determine which with certainty?

**Projective measurement**: No. Non-orthogonal states cannot be perfectly distinguished by projective measurement (overlapping eigenspaces).

**POVM solution**: Yes, with a trade-off. Use a **three-outcome POVM**:
- Outcome 1: Definitely state $|\psi_1\rangle$.
- Outcome 2: Definitely state $|\psi_2\rangle$.
- Outcome 3: Inconclusive (cannot decide).

For two non-orthogonal states:

:::{admonition} Construction: State Discrimination POVM
:class: dropdown note

For states $|\psi_1\rangle, |\psi_2\rangle$ with overlap $\langle\psi_1|\psi_2\rangle = c$ (with $0 < |c| < 1$):

Define POVM elements:

$$


M_1 = \frac{1}{1+|c|^2}(I - |\psi_2\rangle\langle\psi_2|) = \frac{1}{1+|c|^2}|\psi_1^\perp\rangle\langle\psi_1^\perp|

$$


$$

M_2 = \frac{1}{1+|c|^2}(I - |\psi_1\rangle\langle\psi_1|) = \frac{1}{1+|c|^2}|\psi_2^\perp\rangle\langle\psi_2^\perp|

$$


$$

M_3 = I - M_1 - M_2

$$


where $|\psi_1^\perp\rangle, |\psi_2^\perp\rangle$ are orthogonal complements.

**Properties**:
- If state is $|\psi_1\rangle$: Outcome 1 with probability $1-|c|^2$; outcome 3 with probability $|c|^2$ (never outcome 2).
- If state is $|\psi_2\rangle$: Outcome 2 with probability $1-|c|^2$; outcome 3 with probability $|c|^2$ (never outcome 1).
- **Unambiguous**: Outcomes 1 and 2 identify state with certainty; outcome 3 means try again.
:::
:::{admonition} Neumark's Theorem
:class: important

Every POVM on Hilbert space $\mathcal{H}$ can be realized as a projective measurement on an extended space $\mathcal{H} \otimes \mathcal{H}_\text{aux}$ (original plus auxiliary system).

**Mathematical form**: For POVM $\{M_i\}$ on $\mathcal{H}$, there exist orthogonal projectors $\{P_i\}$ on $\mathcal{H} \otimes \mathcal{H}_\text{aux}$ such that:

$$

M_i = \text{Tr}_{\text{aux}}(P_i (\rho \otimes |0\rangle\langle 0|_\text{aux}))

$$

**Physical interpretation**: POVM is equivalent to entangling with an ancilla, performing a projective measurement on the combined system, and discarding the ancilla outcome.

:::


**Theorem (Neumark)**: Every POVM on Hilbert space $\mathcal{H}$ can be realized as a **projective measurement** on an **extended space** $\mathcal{H} \otimes \mathcal{H}_\text{aux}$ (original plus auxiliary system), followed by tracing out the auxiliary system.

**Mathematical statement**: For POVM $\{M_i\}$ on $\mathcal{H}$, there exist:
- Hilbert space $\mathcal{H}_\text{aux}$ (auxiliary/ancilla system).
- Orthogonal projectors $\{P_i\}$ on $\mathcal{H} \otimes \mathcal{H}_\text{aux}$.

such that:

$$


M_i = \text{Tr}_\text{aux}(P_i (\rho \otimes |0\rangle\langle 0|_\text{aux}))

$$ (eq-povm-neumark)

**Physical interpretation**: POVM is equivalent to entangling with an ancilla, performing a projective measurement on the combined system, and discarding the ancilla outcome.

Neumark's theorem provides a minimal realization:

:::{admonition} Proof: Neumark's Theorem - Minimal Realization
:class: dropdown note

The minimal auxiliary space dimension is $\dim(\mathcal{H}_\text{aux}) = n$ (number of POVM outcomes). For $d$-dimensional original space and $n$ outcomes: auxiliary needs dimension $\approx n - d$.

:::

#:::{admonition} Example: SIC-POVM (Six-State)
:class: example


For a qubit ($d=2$), define 6 POVM elements:

$$


M_k = \frac{1}{3}|\psi_k\rangle\langle\psi_k| \quad (k=1,\ldots,6)

$$


where $|\psi_k\rangle$ are 6 equally-spaced directions on the Bloch sphere (vertices of icosahedron).

**Properties**:
- $\sum_k M_k = I$ (completeness verified by symmetry).
- **Informationally complete**: Can reconstruct any qubit state from probabilities $\{p_k\}$.
- More outcomes ($n=6$) than dimensions ($d=2$); overcomplete.

:::
:::{admonition} Example: Heterodyne Detection POVM
:class: example


In quantum optics, measuring quadrature amplitudes: POVM elements are **Gaussian** states (coherent states), indexed by phase space points.

$$


M(\alpha) = \frac{1}{\pi}|\alpha\rangle\langle\alpha| \quad (\alpha \in \mathbb{C})

$$


**Properties**:
- Continuous outcome space (not discrete).
- Each outcome has overlap with many states (not orthogonal).
- Allows full quantum tomography of optical fields.

:::
### Square Roots of POVM Elements

The post-measurement state involves $\sqrt{M_i}$ (square root operator). Since $M_i \geq 0$ (positive semidefinite), the square root is well-defined via spectral decomposition:

$$


M_i = \sum_j \lambda_{i,j} |v_{i,j}\rangle\langle v_{i,j}| \quad \Rightarrow \quad \sqrt{M_i} = \sum_j \sqrt{\lambda_{i,j}} |v_{i,j}\rangle\langle v_{i,j}|

$$


**Physical interpretation**: $\sqrt{M_i}$ represents the "measurement operator" (Kraus operator) associated with outcome $i$; $M_i = \sqrt{M_i}^\dagger \sqrt{M_i}$ is the "effect operator".

### Weak Measurement

A **weak measurement** is a limiting case where the measurement extracts partial information, minimizing state disturbance. It can be realized via:
- Weak coupling to an ancilla.
- POVM with "soft" (gradual) state collapse.
- Post-selection on weak measurement outcome.

**Allows**: Feedback control, sequential measurements, detecting small effects without strong collapse.

### Informationally Complete POVMs (IC-POVM)

A POVM is **informationally complete** if the probabilities $\{p_i = \text{Tr}(M_i \rho)\}$ uniquely determine the state $\rho$.

**Minimum size**: At least $d^2$ outcomes needed for a $d$-dimensional system (to uniquely specify $d^2$ real parameters of $\rho$).

**Symmetric IC-POVM (SIC-POVM)**: Special case with maximal symmetry and minimal overlap. For qubits, 4 outcomes suffice (tetrahedron vertices on Bloch ball).

**Application**: Full quantum state tomography with minimal measurements; important for quantum certification and characterization.

### Connection to Quantum Channels

A POVM outcome describes a **quantum channel** in the sense of Kraus operators:

$$


\mathcal{E}(\rho) = \sum_i \sqrt{M_i} \rho \sqrt{M_i}

$$


where the sum is over all outcomes (unconditional), averaging the post-measurement states.

The Kraus operators are $K_i = \sqrt{M_i}$, and they satisfy:

$$


\sum_i K_i^\dagger K_i = \sum_i M_i = I

$$


(trace-preserving condition for a channel).

:::{admonition} POVM vs projective
:class: important
A POVM $\{M_i\}$ with $\sum_i M_i = I$ and $M_i \geq 0$ generalizes projective measurement: the elements need not be orthogonal or rank-1. POVMs can have more outcomes than the Hilbert space dimension, enabling tasks impossible with projective measurements (e.g., unambiguous state discrimination).
:::

### Summary

- **POVM**: Set of positive operators $\{M_i\}$ with $M_i \geq 0$ and $\sum_i M_i = I$; generalizes orthogonal projectors.
- **Measurement**: Outcome $i$ with probability $p_i = \text{Tr}(M_i \rho)$; post-measurement state $\rho' = \sqrt{M_i}\rho\sqrt{M_i}/p_i$.
- **Advantages**: Non-orthogonal elements; more outcomes than dimensions; unambiguous state discrimination.
- **Unambiguous discrimination**: Three-outcome POVM can distinguish non-orthogonal states with inconclusive outcome (trade-off).
- **Neumark's theorem**: POVM on $\mathcal{H}$ equals projective measurement on $\mathcal{H} \otimes \mathcal{H}_\text{aux}$ (extended space with auxiliary system).
- **Examples**: 6-state POVM (qubits), heterodyne detection (optics), SIC-POVM (tomography).
- **IC-POVM**: Informationally complete POVM uniquely determines state from probabilities; needs $\geq d^2$ outcomes.
