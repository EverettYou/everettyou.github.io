## Lecture Notes

Entanglement admits rich quantitative structure. This section develops tools to quantify and detect entanglement.

### Overview

A quantum state can be assigned a numerical **entanglement measure**: a function that vanishes on separable states and ranges to a maximum for maximally entangled states. Key measures:

- **Entanglement entropy**: information-theoretic measure for pure bipartite states.
- **Concurrence**: computable entanglement measure for two-qubit states (pure and mixed).
- **Entanglement witnesses**: Hermitian operators that certify entanglement without full tomography.
- **Partial transpose and negativity**: Peres–Horodecki criterion for entanglement detection.

### Entanglement Entropy (Recap)

For a bipartite pure state $\vert \Psi\rangle_{AB}$ with Schmidt decomposition $\vert \Psi\rangle = \sum_k \lambda_k \vert u_k\rangle_A \vert v_k\rangle_B$:

$$
E(\Psi) = S(\hat{\rho}_A) = -\sum_k \lambda_k^2 \ln \lambda_k^2
$$ (eq-ent-entropy)

where $\hat{\rho}_A = \mathrm{Tr}_B(\vert \Psi\rangle\langle\Psi\vert)$.

:::{admonition} Properties of Entanglement Entropy
:class: note

- **Vanishes for product states**: $E = 0$ if $\vert \Psi\rangle = \vert \psi\rangle_A \otimes \vert \phi\rangle_B$.
- **Maximized for maximally entangled states**: Bell states achieve $E = \ln 2$.
- **Symmetric**: $S(\hat{\rho}_A) = S(\hat{\rho}_B)$.
- **LOCC monotone**: Cannot increase under local operations and classical communication.
:::

:::{admonition} Example: Partially Entangled State
:class: dropdown example

**Problem.** Consider $\vert \Psi\rangle = \cos\theta\, \vert 00\rangle + \sin\theta\, \vert 11\rangle$. Find the entanglement entropy.

**Solution.** Schmidt coefficients: $\lambda_1 = \cos\theta$, $\lambda_2 = \sin\theta$.

$$
E(\theta) = -\cos^2\theta\, \ln(\cos^2\theta) - \sin^2\theta\, \ln(\sin^2\theta)
$$

At $\theta = 0$: $E = 0$ (product). At $\theta = \pi/4$: $E = \ln 2$ (Bell state).

:::


:::{admonition} Poll: Entanglement entropy for pure states
:class: dropdown poll

For a pure bipartite state $|\psi\rangle_{AB}$, the entanglement entropy is $E(\psi) = S(\rho_A) = -\operatorname{Tr}(\rho_A \log_2 \rho_A)$, where $\rho_A = \operatorname{Tr}_B(|\psi\rangle\langle\psi|)$. When is $E(\psi) = 0$?

(A) Only when $|\psi\rangle = |a\rangle_A \otimes |b\rangle_B$ (product state).

(B) When both subsystems are in pure states.

(C) When the total state has zero entropy.

(D) Never — all entangled states have $E(\psi) > 0$.

:::

### Rényi Entanglement Entropies

Entanglement entropy captures the *average* information about subsystem $A$ that is lost when we discard subsystem $B$. But does the full distribution of entanglement matter? **Rényi entropies generalize this picture**: instead of a single average, they probe the entire spectrum of Schmidt coefficients. The limit $\alpha \to \infty$ isolates the single largest coefficient—revealing the least-mixed aspect of the state.

The Rényi-$\alpha$ entanglement entropy:

$$
E_\alpha(\Psi) = \frac{1}{1-\alpha} \ln \sum_k \lambda_k^{2\alpha}
$$ (eq-ent-renyi)

- $\alpha \to 1$: recovers von Neumann entropy.
- $\alpha = 2$: $E_2 = -\ln \mathrm{Tr}(\hat{\rho}_A^2)$, experimentally accessible via "swap trick".
- $\alpha = 0$: $E_0 = \ln(\text{Schmidt rank})$.

### Bell States

:::{admonition} Definition: Bell States
:class: important

$$
\vert \Phi^+\rangle = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle), \quad \vert \Phi^-\rangle = \frac{1}{\sqrt{2}}(\vert 00\rangle - \vert 11\rangle)
$$

$$
\vert \Psi^+\rangle = \frac{1}{\sqrt{2}}(\vert 01\rangle + \vert 10\rangle), \quad \vert \Psi^-\rangle = \frac{1}{\sqrt{2}}(\vert 01\rangle - \vert 10\rangle)
$$

These four states are **mutually orthogonal**, form a **basis** for $\mathbb{C}^2 \otimes \mathbb{C}^2$, and are **maximally entangled**: $E = \ln 2$ and $\hat{\rho}_A = I/2$ on each subsystem.
:::

:::{admonition} Discussion
:class: dropdown tip

Bell states are maximally entangled, yet their reduced density matrices are completely mixed. Does a mixed reduced state always signal entanglement? Can a separable state have a mixed reduced density matrix on one subsystem? (Hint: think about a product of mixed states.)
:::

### Concurrence (Two Qubits)

Entanglement entropy works perfectly for pure states, but breaks down for mixed states: a mixed **separable** state can have nonzero entropy even though it carries no entanglement. For mixed states, we need a measure that cleanly distinguishes quantum entanglement from classical correlations. **Concurrence** fills this gap for two-qubit systems, vanishing exactly on separable states and reaching unity for maximally entangled Bell states.

:::{admonition} Definition: Concurrence (Pure States)
:class: important

For a pure two-qubit state $\vert \Psi\rangle = \sum_{ij} c_{ij}\vert ij\rangle$:

$$
\mathcal{C}(\vert \Psi\rangle) = 2|c_{00}c_{11} - c_{01}c_{10}\vert
$$ (eq-ent-concurrence-pure)

$\mathcal{C} = 0$ for separable states; $\mathcal{C} = 1$ for maximally entangled (Bell) states.
:::

:::{admonition} Example: Concurrence of Bell States
:class: dropdown example

**Problem.** Compute the concurrence for $\vert \Phi^+\rangle = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$ and the product state $\vert 01\rangle$.

**Solution.** For $\vert \Phi^+\rangle$: $\mathcal{C} = 1$. For $\vert 01\rangle$: $\mathcal{C} = 0$.

:::

**Wootters' Concurrence for Mixed States.**

Construct the spin-flipped density matrix:

$$
\tilde{\hat{\rho}} = (\hat{\sigma}^y \otimes \hat{\sigma}^y) \hat{\rho}^* (\hat{\sigma}^y \otimes \hat{\sigma}^y)
$$ (eq-ent-spin-flip)

:::{admonition} Wootters' Concurrence
:class: important

Compute eigenvalues $\mu_1 \geq \mu_2 \geq \mu_3 \geq \mu_4$ of $\sqrt{\sqrt{\hat{\rho}} \tilde{\hat{\rho}} \sqrt{\hat{\rho}}}$:

$$
\mathcal{C}(\hat{\rho}) = \max(0, \mu_1 - \mu_2 - \mu_3 - \mu_4)
$$ (eq-ent-concurrence-mixed)

$\mathcal{C} \in [0, 1]$; $\mathcal{C} = 0$ iff separable; $\mathcal{C} = 1$ iff Bell state. Related to entanglement of formation: $E_f(\hat{\rho}) = h\left(\frac{1 + \sqrt{1 - \mathcal{C}^2}}{2}\right)$ where $h(x) = -x\ln x - (1-x)\ln(1-x)$.
:::

### Entanglement Witnesses

Computing entanglement measures requires either full state tomography (expensive) or specialized matrix eigenvalue algorithms. **Witnesses** offer a shortcut: a single Hermitian operator whose negative expectation value certifies entanglement directly, without reconstructing the full state.

:::{admonition} Definition: Entanglement Witness
:class: important

$W$ is an entanglement witness if:

- $\mathrm{Tr}(W \hat{\rho}_{\text{sep}}) \geq 0$ for all separable states.
- $\mathrm{Tr}(W \hat{\rho}_{\text{ent}}) < 0$ for at least one entangled state.

$\langle W \rangle < 0$ certifies entanglement without full tomography.
:::

:::{admonition} Example: Witness for Bell States
:class: dropdown example

**Problem.** Construct a witness for $\vert \Phi^+\rangle$.

**Solution.** Define $W = \frac{I}{2} - \vert \Phi^+\rangle\langle\Phi^+\vert$. For separable states, $\langle\Phi^+\vert\hat{\rho}_{\text{sep}}\vert \Phi^+\rangle \leq \frac{1}{2}$, so $\mathrm{Tr}(W\hat{\rho}_{\text{sep}}) \geq 0$. For $\vert \Phi^+\rangle$: $\mathrm{Tr}(W\vert \Phi^+\rangle\langle\Phi^+\vert) = -\frac{1}{2} < 0$.

:::

:::{admonition} Entanglement Witness Completeness
:class: important

Every entangled state admits a witness (Hahn–Banach). However, no single witness detects all entangled states.
:::

### Hierarchy of Quantum Correlations

The **partial transpose** is a linear map that flips indices on one subsystem. For a 2-qubit state $\hat{\rho} = \sum_{ij,kl} \hat{\rho}_{ij,kl} \vert ij\rangle\langle kl\vert$, applying partial transpose on subsystem $B$ yields: $\hat{\rho}^{T_B} \mapsto \vert il\rangle\langle kj\vert$ (the $B$-indices swap in the bra and ket). This simple operation reveals deep structure: a negative eigenvalue signals entanglement.

:::{admonition} Definition: Partial Transpose and PPT States
:class: important

The **partial transpose** with respect to subsystem B:

$$
\hat{\rho}^{T_B} = \sum_{ijkl} \hat{\rho}_{ij,kl} \vert ik\rangle\langle jl\vert
$$

A state is **PPT** if $\hat{\rho}^{T_B} \geq 0$. **Peres–Horodecki Criterion**: If $\hat{\rho}^{T_B}$ has a negative eigenvalue, then $\hat{\rho}$ is entangled.
:::

:::{admonition} Definition: Negativity
:class: important

$$
\mathcal{N}(\hat{\rho}) = \frac{\|\hat{\rho}^{T_B}\|_1 - 1}{2}
$$ (eq-ent-negativity)

$\mathcal{N} = 0$ for PPT states; $\mathcal{N} > 0$ certifies entanglement.
:::

:::{admonition} Example: Bell State Partial Transpose
:class: dropdown example

**Problem.** Compute $\hat{\rho}^{T_B}$ for $\vert \Phi^+\rangle = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$.

**Solution.** Eigenvalues of $\hat{\rho}^{T_B}$: $1/2, 1/2, -1/2, -1/2$. Negative eigenvalues certify entanglement.

:::

:::{admonition} Entanglement Classification Hierarchy
:class: note

$$
\text{Separable} \subset \text{PPT} \subset \text{All States}
$$

- **Separable states**: $\hat{\rho} = \sum_i p_i \hat{\rho}_A^i \otimes \hat{\rho}_B^i$ (no entanglement).
- **PPT states**: Positive partial transpose; includes separable states plus entangled states in $2 \times 2$ and $2 \times 3$ systems.
- **NPT entangled states**: Negative partial transpose; definitely entangled.

For $2 \times 2$ and $2 \times 3$ systems: **PPT $\Leftrightarrow$ separable** (Peres conjecture, proved). For higher dimensions, **bound entangled** states exist—entangled but PPT, undetectable by partial transpose alone.
:::

### Summary

- **Entanglement entropy:** For pure bipartite state, $E(\vert\psi\rangle) = S(\operatorname{Tr}_B \vert\psi\rangle\langle\psi\vert)$ (von Neumann entropy of subsystem); zero iff product state, maximal for maximally entangled Bell states.
- **Concurrence:** $C(\hat{\rho}) = \max(0, \lambda_1 - \lambda_2 - \lambda_3 - \lambda_4)$ (for 2-qubit systems) ranges 0 (separable) to 1 (maximally entangled); related to entanglement of formation.
- **Separability:** State $\hat{\rho}$ separable iff $\hat{\rho} = \sum_i p_i \hat{\rho}_A^{(i)} \otimes \hat{\rho}_B^{(i)}$; not entangled. Checking separability is NP-hard for large systems.
- **Entanglement of formation:** Minimum average entanglement needed to create $\hat{\rho}$ from product states; for pure states, equals entanglement entropy; for mixed states, optimization over decompositions.
- **Monogamy of entanglement:** Entanglement cannot be freely shared—if $A$ highly entangled with $B$, it cannot be highly entangled with $C$; quantified by subadditivity of entanglement measures.
