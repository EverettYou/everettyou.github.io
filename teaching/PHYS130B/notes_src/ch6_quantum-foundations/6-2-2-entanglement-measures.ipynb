## Lecture Notes

Entanglement is not merely a binary property (entangled or not)—it admits a rich quantitative structure. How entangled is a given state? Can we assign a number to it? The answers draw on information theory, geometry, and operator theory, and have deep implications for quantum computing, communication, and the structure of many-body physics.

### Entanglement Entropy



:::{admonition} Discussion
:class: dropdown warning

Entanglement entropy and entanglement of formation agree for pure states but not mixed states. Construct a 2-qubit mixed state where E_F ≠ S(ρ_A) and compute both.
:::


:::{admonition} Definition: Entanglement Entropy
:class: important

For a bipartite pure state $|\Psi\rangle_{AB}$ with Schmidt decomposition $|\Psi\rangle = \sum_k \lambda_k |u_k\rangle_A |v_k\rangle_B$, the **entanglement entropy** is defined as the von Neumann entropy of the reduced density matrix:

$$


E(\Psi) = S(\rho_A) = -\text{Tr}(\rho_A \ln \rho_A) = -\sum_k \lambda_k^2 \ln \lambda_k^2

$$ (eq-ent-measure-entropy)

where $\rho_A = \text{Tr}_B(|\Psi\rangle\langle\Psi|)$ has eigenvalues $\lambda_k^2$ (the squared Schmidt coefficients from §6.2.1).
:::

Key properties:

- **Vanishes for product states.** If $|\Psi\rangle = |\psi\rangle_A \otimes |\phi\rangle_B$, then $\rho_A$ is pure and $E = 0$.
- **Maximized for maximally entangled states.** For two qubits, the Bell states achieve $E = \ln 2$ (1 ebit).
- **Symmetric.** $S(\rho_A) = S(\rho_B)$ for any bipartite pure state—a remarkable consequence of the Schmidt decomposition.
- **Monotone.** Cannot increase under local operations and classical communication (LOCC).

:::{admonition} Example: Bell State Entanglement Entropy
:class: example

For the Bell state $|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$, the reduced density matrix is:

$$


\rho_A = \frac{1}{2}|0\rangle\langle 0| + \frac{1}{2}|1\rangle\langle 1| = \frac{I}{2}

$$


Eigenvalues: $\lambda_1^2 = \lambda_2^2 = 1/2$. The entanglement entropy is:

$$


E = -\frac{1}{2}\ln\frac{1}{2} - \frac{1}{2}\ln\frac{1}{2} = \ln 2 \approx 0.693 \text{ nats}

$$


This is **maximal** for two qubits: one ebit of entanglement.
:::

:::{admonition} Example: Partially Entangled State
:class: example

Consider $|\Psi\rangle = \cos\theta\, |00\rangle + \sin\theta\, |11\rangle$ with $0 < \theta < \pi/4$.

Schmidt coefficients: $\lambda_1 = \cos\theta$, $\lambda_2 = \sin\theta$. The entanglement entropy is:

$$


E(\theta) = -\cos^2\theta\, \ln(\cos^2\theta) - \sin^2\theta\, \ln(\sin^2\theta)

$$


- At $\theta = 0$: product state, $E = 0$.
- At $\theta = \pi/4$: Bell state, $E = \ln 2$ (maximum).
- For intermediate $\theta$: $0 < E < \ln 2$ (partial entanglement).

The entanglement entropy interpolates smoothly between separable and maximally entangled.
:::

### Rényi Entanglement Entropies

The Rényi-$\alpha$ entanglement entropy generalizes entanglement entropy:

$$


E_\alpha(\Psi) = S_\alpha(\rho_A) = \frac{1}{1-\alpha} \ln \text{Tr}(\rho_A^\alpha) = \frac{1}{1-\alpha} \ln \sum_k \lambda_k^{2\alpha}

$$ (eq-ent-measure-renyi)

- $\alpha \to 1$: recovers von Neumann entanglement entropy.
- $\alpha = 2$: Rényi-2 entropy $E_2 = -\ln \text{Tr}(\rho_A^2) = -\ln(\text{purity})$, experimentally accessible via randomized measurements without full tomography.
- $\alpha = 0$: $E_0 = \ln(\text{Schmidt rank})$, counts the dimension of entanglement.

:::{note}
Rényi-2 entropy is particularly important in experiments because $\text{Tr}(\rho_A^2)$ can be measured by performing identical measurements on two copies of the state (the "swap trick"), avoiding the need for full quantum state tomography.
:::

### Concurrence (Two Qubits)

For two-qubit systems, a computable entanglement measure is the **concurrence** $\mathcal{C}$.

#### Pure States

:::{admonition} Definition: Concurrence (Pure Two-Qubit States)
:class: important

For a pure two-qubit state $|\Psi\rangle = \sum_{ij} c_{ij}|ij\rangle$:

$$


\mathcal{C}(|\Psi\rangle) = 2|\det(C)| = 2|c_{00}c_{11} - c_{01}c_{10}|

$$ (eq-ent-measure-concurrence-pure)

where $C = [c_{ij}]$ is the $2\times 2$ coefficient matrix.

- $\mathcal{C} = 0$: separable (product state).
- $\mathcal{C} = 1$: maximally entangled (Bell state).
:::

:::{admonition} Example: Concurrence of a General Two-Qubit State
:class: example

For $|\Psi\rangle = \alpha|00\rangle + \beta|11\rangle$ (with $|\alpha|^2 + |\beta|^2 = 1$):

$$


C = \begin{pmatrix} \alpha & 0 \\ 0 & \beta \end{pmatrix}, \quad \mathcal{C} = 2|\alpha\beta|

$$


- $\alpha = 1, \beta = 0$: $\mathcal{C} = 0$ (product state $|00\rangle$).
- $\alpha = \beta = 1/\sqrt{2}$: $\mathcal{C} = 1$ (Bell state $|\Phi^+\rangle$).
- $\alpha = \sqrt{3}/2, \beta = 1/2$: $\mathcal{C} = \sqrt{3}/2 \approx 0.87$ (partially entangled).
:::

#### Mixed States (Wootters' Formula)

For mixed two-qubit states, **Wootters' concurrence** is one of the few computable entanglement measures. It requires computing a "spin-flipped" version of the density matrix:

$$


\tilde{\rho} = (\hat{\sigma}^y \otimes \hat{\sigma}^y) \rho^* (\hat{\sigma}^y \otimes \hat{\sigma}^y)

$$ (eq-ent-measure-spin-flip)

where $\rho^*$ is complex conjugation in the computational basis, and $\hat{\sigma}^y$ are Pauli matrices.

:::{admonition} Wootters' Concurrence for Mixed States
:class: important

Compute the eigenvalues $\mu_1 \geq \mu_2 \geq \mu_3 \geq \mu_4$ of the matrix product $\sqrt{\sqrt{\rho} \tilde{\rho} \sqrt{\rho}}$, and evaluate:

$$


\mathcal{C}(\rho) = \max(0, \mu_1 - \mu_2 - \mu_3 - \mu_4)

$$ (eq-ent-measure-concurrence-mixed)

**Properties**:
- $\mathcal{C}(\rho) \in [0, 1]$ for two-qubit systems
- $\mathcal{C} = 0$ iff the state is separable
- $\mathcal{C} = 1$ iff the state is maximally entangled (Bell state)
- This formula is one of the few **computable** entanglement measures for mixed states.
:::

:::{admonition} Derivation: Wootters' Concurrence Formula
:class: dropdown note

**Construction of the formula:**

The concurrence for a pure state was: $\mathcal{C}(|\Psi\rangle) = 2|c_1 c_2|$ (product of Schmidt coefficients scaled).

For mixed states, Wootters generalized this by introducing an auxiliary "spin-flipped" version:

$$


\tilde{\rho} = (\hat{\sigma}^y \otimes \hat{\sigma}^y) \rho^* (\hat{\sigma}^y \otimes \hat{\sigma}^y)

$$


The idea: $\tilde{\rho}$ transforms the state in a specific way related to its entanglement structure. 

**Key steps**:

1. Compute $R = \sqrt{\sqrt{\rho} \tilde{\rho} \sqrt{\rho}}$ (the geometric mean of $\rho$ and $\tilde{\rho}$).

2. Find eigenvalues $\mu_j$ of $R$ in decreasing order.

3. Apply the formula: $\mathcal{C}(\rho) = \max(0, \mu_1 - \mu_2 - \mu_3 - \mu_4)$.

The max with 0 ensures concurrence is non-negative; separable states automatically have $\mathcal{C} = 0$.

:::
#### Relation to Entanglement of Formation

The **entanglement of formation** $E_f$ (minimum average entanglement over all pure-state decompositions of $\rho$) is related to concurrence by:

$$


E_f(\rho) = h\!\left(\frac{1 + \sqrt{1 - \mathcal{C}^2}}{2}\right)

$$ (eq-ent-measure-eof)

where $h(x) = -x\ln x - (1-x)\ln(1-x)$ is the binary entropy function.

### Entanglement Witnesses

An **entanglement witness** is a Hermitian operator $W$ that detects entanglement without full state tomography.

#### Definition

:::{admonition} Definition: Entanglement Witness
:class: important

$W$ is an entanglement witness if:

- $\text{Tr}(W \rho_{\text{sep}}) \geq 0$ for all separable states $\rho_{\text{sep}}$.
- $\text{Tr}(W \rho_{\text{ent}}) < 0$ for at least one entangled state $\rho_{\text{ent}}$.

A negative expectation value $\langle W \rangle < 0$ **certifies** that the state is entangled.
:::

A systematic construction of entanglement witnesses:

:::{admonition} Construction: Entanglement Witnesses from Bell States
:class: dropdown note

For the Bell state $|\Phi^+\rangle$, a witness is:

$$


W = \frac{1}{2}I - |\Phi^+\rangle\langle\Phi^+|

$$ (eq-ent-measure-witness)

- For any separable state: $\langle\Phi^+|\rho_{\text{sep}}|\Phi^+\rangle \leq 1/2$, so $\text{Tr}(W\rho_{\text{sep}}) \geq 0$.
- For $\rho = |\Phi^+\rangle\langle\Phi^+|$: $\text{Tr}(W\rho) = 1/2 - 1 = -1/2 < 0$.

:::{admonition} Example: Detecting Entanglement with a Witness
:class: example

Consider the noisy Bell state $\rho = p |\Phi^+\rangle\langle\Phi^+| + (1-p)\frac{I}{4}$ (Werner state).

Using the witness $W = \frac{1}{2}I - |\Phi^+\rangle\langle\Phi^+|$:

$$


\text{Tr}(W\rho) = \frac{1}{2} - p - (1-p)\frac{1}{4} = \frac{1}{2} - p - \frac{1}{4} + \frac{p}{4} = \frac{1}{4} - \frac{3p}{4}

$$


This is negative when $p > 1/3$. So the witness detects entanglement for $p > 1/3$, even though the state is mixed with significant noise.
:::

:::{important}
By the Hahn–Banach theorem, **every entangled state can be detected by some witness**. However, no single witness detects all entangled states—different states require different witnesses. Finding optimal witnesses is an active area of research.
:::
:::
### Negativity

:::{admonition} Definition: Negativity
:class: important

The **negativity** exploits the **partial transpose** criterion (Peres–Horodecki):

$$


\mathcal{N}(\rho) = \frac{\|\rho^{T_B}\|_1 - 1}{2}

$$ (eq-ent-measure-negativity)

where $\rho^{T_B}$ is the partial transpose with respect to subsystem B, and $\|\cdot\|_1$ is the trace norm ($\sum_i |\lambda_i|$).

- $\mathcal{N} = 0$: state has positive partial transpose (PPT); separable for $2 \times 2$ and $2 \times 3$ systems.
- $\mathcal{N} > 0$: state is entangled (negative eigenvalue under partial transpose).
- **Logarithmic negativity** $E_N = \ln\|\rho^{T_B}\|_1$ is also widely used.
:::

:::{caution}
In dimensions $\geq 3 \times 3$, there exist entangled states with positive partial transpose (PPT entangled states, or "bound entangled" states). Negativity fails to detect these; additional criteria are needed.
:::

### Hierarchy of Entanglement Measures

Different measures capture different aspects of entanglement:

| Measure | Applies to | Computable? | Operational meaning |
|---------|-----------|-------------|---------------------|
| Entanglement entropy $E$ | Pure bipartite | Yes | Distillable entanglement (asymptotic) |
| Concurrence $\mathcal{C}$ | Two-qubit (pure & mixed) | Yes | Related to entanglement of formation |
| Negativity $\mathcal{N}$ | Any bipartite | Yes | Entanglement cost under PPT operations |
| Entanglement of formation $E_f$ | Any bipartite | Only for 2 qubits | Minimum entanglement to create state |
| Distillable entanglement $E_d$ | Any bipartite | Generally no | Maximum Bell pairs extractable |
| Squashed entanglement $E_{sq}$ | Any bipartite | Generally no | Faithful and additive |

:::{note}
All valid entanglement measures must satisfy: (1) vanish on separable states, (2) not increase under LOCC (monotonicity), and (3) reduce to entanglement entropy on pure states. These axioms define an **entanglement monotone**.
:::

### Summary

- **Entanglement entropy** $E = S(\rho_A)$: fundamental measure for pure bipartite states; ranges from 0 (separable) to $\ln d$ (maximally entangled).
- **Rényi entropies** $E_\alpha$: generalize entanglement entropy; $E_2$ is experimentally accessible via the swap trick.
- **Concurrence** $\mathcal{C}$: computable for two-qubit states (pure and mixed); ranges from 0 to 1.
- **Entanglement witnesses**: operators detecting entanglement via $\langle W \rangle < 0$; every entangled state has a witness (Hahn–Banach).
- **Negativity**: uses partial transpose; computable but misses bound entangled states in high dimensions.
- **No single measure suffices**: different measures capture distinct operational aspects (distillation, formation, detection).
