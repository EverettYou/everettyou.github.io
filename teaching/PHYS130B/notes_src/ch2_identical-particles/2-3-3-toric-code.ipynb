## Lecture Notes

### Overview

The toric code demonstrates the most remarkable consequence of 2D topology: **particles can emerge from qubits**. Starting from a lattice of spin-1/2 qubits with local interactions, the model produces two types of anyonic excitations — charges ($e$) and fluxes ($m$) — whose composite $\varepsilon = e \times m$ is a **fermion**, even though the microscopic ingredients are purely bosonic. This lesson connects the emergent anyon statistics back to the charge-flux composite picture of §2.3.1 and ends with the toric code as a topological quantum error-correcting code.

### From Qubits to Particles

One of the most remarkable facts in modern physics is that **particles can emerge from qubits**. The toric code, introduced by Alexei Kitaev in 2003, is the simplest exactly solvable model demonstrating this. Starting from a lattice of spin-1/2 qubits with purely local interactions, the model produces emergent excitations that behave as anyons — including objects that are effectively **bosons** and **fermions**, even though the microscopic degrees of freedom are just qubits on lattice edges.

### Toric Code Model

Place a qubit on every **edge** of a square lattice on a torus. The model is built from two local operators: one associated with each vertex and one with each plaquette.

:::{admonition} Toric Code Hamiltonian
:class: important

$$
\hat{H} = -\sum_v \hat{A}_v - \sum_p \hat{B}_p
$$

- $\hat{A}_v$ (**star term**): multiplies $\hat{X}$ on the four edges touching vertex $v$.

$$
\hat{A}_v = \prod_{e \ni v} \hat{X}_e
$$

- $\hat{B}_p$ (**plaquette term**): multiplies $\hat{Z}$ on the four edges around plaquette $p$.

$$
\hat{B}_p = \prod_{e \in p} \hat{Z}_e
$$
:::

All terms in the Hamiltonian commute:
$[\hat{A}_v, \hat{A}_{v'}] = [\hat{B}_p, \hat{B}_{p'}] = [\hat{A}_v, \hat{B}_p] = 0$.
Hence the ground state is the simultaneous $+1$ eigenspace of all stars and plaquettes: $A_v = +1$, $B_p = +1$ for every $v,p$.
:::{admonition} Derivation: All Terms in the Toric Code Commute
:class: dropdown information

We verify $[\hat{A}_v, \hat{B}_p] = 0$ for any vertex $v$ and plaquette $p$.

**Case 1: $v$ and $p$ share no edges.** Then $\hat{A}_v$ and $\hat{B}_p$ act on completely disjoint sets of qubits. Operators on different qubits trivially commute, so $[\hat{A}_v, \hat{B}_p] = 0$.

**Case 2: $v$ and $p$ share edges.** A vertex touches 4 edges; a plaquette borders 4 edges. By the geometry of the square lattice, a vertex $v$ and an adjacent plaquette $p$ always share **exactly 2 edges** (call them $e_1$ and $e_2$).

On each shared edge, $\hat{A}_v$ contributes an $\hat{X}$ and $\hat{B}_p$ contributes a $\hat{Z}$. Since $\hat{X}$ and $\hat{Z}$ anticommute on a single qubit ($\hat{X}\hat{Z} = -\hat{Z}\hat{X}$), each shared edge contributes a factor of $(-1)$ when we commute $\hat{A}_v$ past $\hat{B}_p$. With two shared edges:

$$
\hat{A}_v \hat{B}_p = (-1)^2 \, \hat{B}_p \hat{A}_v = \hat{B}_p \hat{A}_v
$$

Therefore $[\hat{A}_v, \hat{B}_p] = 0$. The same parity argument shows $[\hat{A}_v, \hat{A}_{v'}] = 0$ (two stars share 0 or 1 edge; $\hat{X}$ commutes with $\hat{X}$) and $[\hat{B}_p, \hat{B}_{p'}] = 0$ (two plaquettes share 0 or 1 edge; $\hat{Z}$ commutes with $\hat{Z}$).

**Key principle:** whenever two Pauli-string operators overlap on an **even** number of qubits where they use anticommuting Pauli matrices, the operators commute. This is the foundation of **stabilizer codes**.
:::


### Two Types of Excitations

Excitations are created by violating either the star or plaquette constraint:

:::{admonition} Charges and Fluxes
:class: important

| Excitation | Violated constraint | Created by | Name |
|------------|-------------------|------------|------|
| **Charge** ($e$) | $A_v = -1$ at vertex $v$ | $Z$-string | Electric charge |
| **Flux** ($m$) | $B_p = -1$ at plaquette $p$ | $X$-string | Magnetic flux |

- Charges are created in **pairs** at the endpoints of a ${Z}$-string
- Fluxes are created in **pairs** at the endpoints of a ${X}$-string
:::

The key to understanding the statistics is the **anticommutation between crossing strings**. A ${Z}$-string (creating $e$ charges) and an ${X}$-loop (detecting $m$ fluxes) share exactly one edge when they cross, and on that edge $\hat{X}\hat{Z} = -\hat{Z}\hat{X}$. Therefore:

:::{admonition} String Anticommutation
:class: important

If an open ${Z}$-string $\hat{S}_e$ crosses a closed ${X}$-loop $\hat{L}_m$ at a single edge:

$$

\hat{S}_e \, \hat{L}_m = -\hat{L}_m \, \hat{S}_e

$$ (eq-string-anticommute)

Moving an $e$ charge around an $m$ flux picks up a phase of $-1$: this is the Aharonov-Bohm phase of a unit charge ($q = 1$) encircling a $\pi$-flux ($\Phi/\hbar = \pi$):

$$

\mathrm{e}^{\mathrm{i}\, q\, \Phi/\hbar} = \mathrm{e}^{\mathrm{i}\pi} = -1

$$
:::

This identifies $e$ as a **unit charge** ($q = 1, \Phi/\hbar = 0$) and $m$ as a **$\pi$-flux** ($q = 0, \Phi/\hbar = \pi$), connecting the toric code to the charge-flux composite picture of §2.3.1.

:::{admonition} Derivation: String Operators Create Deconfined Anyons at Their Endpoints
:class: dropdown information

We show that a $\hat{Z}$-string operator $\hat{S}^Z_{v_1 v_2}$ creates exactly two $e$-anyons at its endpoints, with an energy cost independent of the string length.

**Definition.** Let $\gamma$ be a path of edges on the lattice from vertex $v_1$ to vertex $v_2$. Define the string operator

$$
\hat{S}^Z_{v_1 v_2} = \prod_{e \in \gamma} \hat{Z}_e
$$

**Effect on star operators.** The star operator at vertex $v$ is $\hat{A}_v = \prod_{e \ni v} \hat{X}_e$. Since $\hat{X}_e$ and $\hat{Z}_e$ anticommute on the same qubit, while $\hat{X}_e$ and $\hat{Z}_{e'}$ commute for $e \neq e'$, we need to count how many edges of $\gamma$ touch each vertex $v$:

- **Interior vertices** of $\gamma$ (neither endpoint): $\gamma$ enters and leaves through exactly 2 edges. Each anticommutes with $\hat{A}_v$, giving $(-1)^2 = +1$. So $[\hat{S}^Z, \hat{A}_v] = 0$ for interior $v$.

- **Endpoint vertices** $v_1, v_2$: only one edge of $\gamma$ touches each endpoint. This gives a factor of $(-1)^1 = -1$, so $\hat{S}^Z \hat{A}_{v_i} = -\hat{A}_{v_i} \hat{S}^Z$.

**Energy eigenvalues.** Apply $\hat{S}^Z$ to the ground state $|\Omega\rangle$ (where all $A_v = +1$):

$$
\hat{A}_{v_i} \hat{S}^Z |\Omega\rangle = -\hat{S}^Z \hat{A}_{v_i} |\Omega\rangle = -\hat{S}^Z |\Omega\rangle
$$

so $\hat{S}^Z|\Omega\rangle$ has $A_{v_1} = A_{v_2} = -1$: exactly two $e$-anyons at the endpoints of $\gamma$.

**Deconfinement.** Each violated star costs energy $+2$ (from $-1 \to +1$ in $-\hat{A}_v$). The total energy cost is $\Delta E = 4$ (two violated stars), **regardless of the length of $\gamma$**. Separating the two anyons by any distance costs no additional energy — they are completely deconfined.

**Consequence.** The anyons can be moved by extending or rerouting the string without changing the energy, since only the endpoints matter. This is the hallmark of topological order: excitations are topological defects, not local perturbations, and their dynamics is governed solely by their positions.
:::


### Anyon Sectors and Statistics

The toric code has four anyon sectors: the vacuum $1$, charge $e$, flux $m$, and the composite $\varepsilon = e \times m$. All their statistics follow from a single fact: the string anticommutation {eq}`eq-string-anticommute`.

**Self-statistics** (exchange phase $\theta_{\mathrm{ex}}$):

- **$e$ is a boson.** Exchanging two $e$ particles involves two $\hat{Z}$-strings. Since $\hat{Z}$-operators commute with each other, no phase is acquired: $\theta_{\mathrm{ex}}(e) = 0$.
- **$m$ is a boson.** Exchanging two $m$ particles involves two $\hat{X}$-strings. Since $\hat{X}$-operators commute with each other: $\theta_{\mathrm{ex}}(m) = 0$.
- **$\varepsilon = e \times m$ is a fermion.** Consider the **topological spin** of $\varepsilon$: rotating a single $\varepsilon$ particle by $2\pi$ is equivalent to moving its $m$-component in a full loop around its $e$-component. This loop threads an $\hat{X}$-string around a $\hat{Z}$-string, and they must cross once, giving a phase of $-1$ by {eq}`eq-string-anticommute`. Therefore the spin phase is $\theta_{\varepsilon} = \pi$. By the spin-statistics relation ($\theta_{\mathrm{ex}} = \theta_{\varepsilon}$), $\varepsilon$ is a **fermion**.
:::{admonition} Derivation: The Composite $\varepsilon = e \times m$ Is a Fermion
:class: dropdown information

We prove that the topological spin of $\varepsilon$ is $\theta_\varepsilon = \pi$, establishing fermionic self-statistics.

**Setup.** An $\varepsilon$ particle at a location consists of an $e$ charge (violated star, $A_v = -1$) and an $m$ flux (violated plaquette, $B_p = -1$) at adjacent vertex and plaquette. The $\varepsilon$ particle is created by a combined $\hat{Z}$-string (for $e$) and $\hat{X}$-string (for $m$) emanating from the same region.

**Topological spin via $2\pi$ rotation.** The topological spin $\theta_s$ of a particle is defined as the phase acquired when the particle is rotated by $2\pi$ about its own axis. For a composite $\varepsilon = e \times m$, a $2\pi$ self-rotation is equivalent to moving the $e$-component in a small loop around the $m$-component (or vice versa).

**Step 1: Braid $e$ around $m$ within the composite.** Move the $e$ charge in a small closed loop encircling the $m$ flux. This requires threading a $\hat{Z}$-string (the worldline of $e$) around an $\hat{X}$-string endpoint (the position of $m$).

**Step 2: Count crossings.** The closed $\hat{Z}$-loop encircling $m$ must cross the $\hat{X}$-string that created $m$ exactly **once** (the string connects $m$ to its partner or extends to the boundary).

**Step 3: Apply string anticommutation.** By {eq}`eq-string-anticommute`, each crossing of a $\hat{Z}$-string and an $\hat{X}$-string contributes a phase of $(-1)$. With one crossing:

$$
\theta_\varepsilon = \pi \quad (\text{phase } = \mathrm{e}^{\mathrm{i}\pi} = -1)
$$

**Step 4: Spin-statistics.** By the spin-statistics relation, the exchange phase equals the topological spin: $\theta_{\mathrm{ex}}(\varepsilon) = \theta_\varepsilon = \pi$. Therefore $\varepsilon$ is a **fermion**.

**Consistency check via the charge-flux picture.** In the charge-flux language, $e$ has $(q=1, \Phi=0)$ and $m$ has $(q=0, \Phi/\hbar = \pi)$. The composite $\varepsilon$ has $(q=1, \Phi/\hbar = \pi)$. The exchange phase is $\theta_{\mathrm{ex}} = q\Phi/(2\hbar) = \pi/2$... but wait, this gives $\pi/2$, not $\pi$! The resolution is that the charge-flux formula $\theta_{\mathrm{ex}} = q\Phi/(2\hbar)$ applies to the mutual statistics contribution. The self-statistics of the composite also includes the mutual braiding of its own constituents during a self-exchange, which adds another $\pi/2$, giving $\theta_{\mathrm{ex}} = \pi/2 + \pi/2 = \pi$.
:::


**Mutual braiding** (phase $\theta_{\mathrm{br}}$ for a full loop):

Taking an $e$ charge around an $m$ flux means threading a $\hat{Z}$-string through an $\hat{X}$-loop. By {eq}`eq-string-anticommute`, each crossing gives $-1$.

:::{admonition} Anyon Statistics Summary
:class: note

**Self-statistics (exchange phase):**

| Sector | $\theta_{\mathrm{ex}}$ | Type |
|--------|----------------------|------|
| $1$ (vacuum) | $0$ | trivial |
| $e$ (charge) | $0$ | boson |
| $m$ (flux) | $0$ | boson |
| $\varepsilon = e \times m$ | $\pi$ | fermion |

**Mutual braiding statistics ($\theta_{\mathrm{br}}$ for a full loop):**

| $\theta_{\mathrm{br}}$ | $1$ | $e$ | $m$ | $\varepsilon$ |
|--|-----|-----|-----|----------------|
| $1$ | $0$ | $0$ | $0$ | $0$ |
| $e$ | $0$ | $0$ | $\pi$ | $\pi$ |
| $m$ | $0$ | $\pi$ | $0$ | $\pi$ |
| $\varepsilon$ | $0$ | $\pi$ | $\pi$ | $0$ |


:::

All entries follow from the $e$-$m$ braiding phase $\theta_{\mathrm{br}}(e,m) = \pi$ and the composition $\varepsilon = e \times m$. For example, braiding $\varepsilon$ around $e$ picks up the $m$-component's phase ($\pi$), and braiding two $\varepsilon$ particles gives $\pi + \pi = 2\pi \equiv 0$.

The remarkable conclusion: **fermions are not fundamental** in the toric code. They emerge as composite objects from a system of qubits that has no fermions in its microscopic description.

### Quantum Error Correction

:::{admonition} Toric Code as a Quantum Error-Correcting Code
:class: important

The toric code is not just a physics toy model — it is one of the most important quantum error-correcting codes:

- **Logical qubits** are encoded in the degenerate ground-state space (4-fold degenerate on a torus)
- **Errors** (bit flips $\hat{X}$ or phase flips $\hat{Z}$) create pairs of excitations
- **Error detection**: measure all $\hat{A}_v$ and $\hat{B}_p$ to locate excitation pairs
- **Error correction**: connect the excitation pairs with minimal strings to annihilate them
- **Topological protection**: a logical error requires a string of errors spanning the entire torus — exponentially unlikely for small error rates
:::
:::{admonition} Derivation: Four-Fold Ground State Degeneracy on the Torus
:class: dropdown information

On a torus with $L \times L$ plaquettes, there are $n = 2L^2$ qubits (edges), $L^2$ star operators $\hat{A}_v$, and $L^2$ plaquette operators $\hat{B}_p$.

**Step 1: Constraints.** The star and plaquette operators satisfy two global constraints:

$$
\prod_{v} \hat{A}_v = \hat{I}, \qquad \prod_p \hat{B}_p = \hat{I}
$$

The first holds because every edge is shared by exactly two vertices, so each $\hat{X}_e$ appears twice in the product and $\hat{X}_e^2 = \hat{I}$. Similarly for plaquettes and $\hat{Z}$. Thus only $L^2 - 1$ star constraints and $L^2 - 1$ plaquette constraints are independent.

**Step 2: Counting stabilizer dimensions.** The total number of independent constraints (stabilizer generators) is $k = 2(L^2 - 1) = 2L^2 - 2$. Each independent stabilizer constraint halves the Hilbert space dimension, so the ground-state subspace (the simultaneous $+1$ eigenspace of all stabilizers) has dimension

$$
\dim(\mathcal{H}_{\mathrm{GS}}) = \frac{2^n}{2^k} = \frac{2^{2L^2}}{2^{2L^2 - 2}} = 2^2 = 4
$$

**Step 3: Logical operators.** The 4-fold degeneracy encodes 2 logical qubits. The logical operators are **non-contractible loops** on the torus:

- $\bar{X}_1$: product of $\hat{X}$ on edges along a horizontal non-contractible cycle
- $\bar{Z}_1$: product of $\hat{Z}$ on edges along a vertical non-contractible cycle
- $\bar{X}_2, \bar{Z}_2$: analogous operators for the other cycle of the torus

These commute with all stabilizers (they are closed loops, so they share an even number of edges with any star or plaquette) but anticommute with each other in the appropriate pairs ($\{\bar{X}_1, \bar{Z}_1\} = 0$), forming the algebra of 2 logical qubits.
:::


The toric code demonstrates the principle of **topological quantum error correction**: information is stored nonlocally in the topology of the ground state, making it immune to any local perturbation. This is the foundation of modern approaches to building fault-tolerant quantum computers using **surface codes** (the planar version of the toric code).


:::{admonition} Derivation: Why the Toric Code Corrects Any Single-Qubit Error
:class: dropdown information

We show that a single $\hat{X}$ or $\hat{Z}$ error on any edge is perfectly detectable and correctable via syndrome measurement.

**Setup.** The code space $\mathcal{C}$ is the simultaneous $+1$ eigenspace of all $\hat{A}_v$ (stars) and $\hat{B}_p$ (plaquettes). A syndrome is detected when any stabilizer measures $-1$.

**Case 1: Single $\hat{Z}$ error on edge $e$.** The operator $\hat{Z}_e$ anticommutes with $\hat{X}_{e'}$ for edges $e' = e$; but $\hat{A}_v = \prod_{e' \ni v} \hat{X}_{e'}$ contains $\hat{X}_e$ only if $v$ is an endpoint of $e$. Edge $e$ has exactly **two** endpoints, call them $v_1$ and $v_2$. Therefore:

$$
\hat{Z}_e \hat{A}_{v_1} = -\hat{A}_{v_1} \hat{Z}_e, \qquad \hat{Z}_e \hat{A}_{v_2} = -\hat{A}_{v_2} \hat{Z}_e,
$$

while $\hat{Z}_e$ commutes with all other stars and with every plaquette ($\hat{B}_p$ is a product of $\hat{Z}$'s only, and $[\hat{Z},\hat{Z}]=0$). Measuring the syndrome after a $\hat{Z}_e$ error reveals exactly $v_1$ and $v_2$ as the two $-1$ vertices. Since vertex $v_1, v_2$ uniquely determine edge $e$, the error location is identified, and applying $\hat{Z}_e$ again restores the code state.

**Case 2: Single $\hat{X}$ error on edge $e$.** By the same argument with stars and plaquettes exchanged: $\hat{X}_e$ anticommutes with $\hat{B}_p$ if and only if $e$ is a boundary edge of plaquette $p$. Edge $e$ borders exactly **two** plaquettes $p_1, p_2$. The syndrome reveals $p_1, p_2$ as the two $-1$ plaquettes, identifying edge $e$; applying $\hat{X}_e$ corrects the error.

**Why logical errors escape detection.** A logical error $\bar{Z}_1 = \prod_{e \in \gamma} \hat{Z}_e$ on a **non-contractible** cycle $\gamma$ is a product of $\hat{Z}$'s that forms a closed loop. Every vertex $v$ either has $0$ or $2$ edges of $\gamma$ incident on it, so $\bar{Z}_1$ commutes with every $\hat{A}_v$—the syndrome is trivial. There is no way to distinguish it from the identity by local syndrome measurements. Correcting such an error requires an operation spanning the entire system, so the logical information is **topologically protected**: local noise cannot cause a logical error.
:::

### Summary

- The **toric code** places qubits on edges of a square lattice with star operators $\hat{A}_v = \prod \hat{X}_e$ and plaquette operators $\hat{B}_p = \prod \hat{Z}_e$.
- Excitations come in two types: **charges** $e$ ($A_v = -1$, created by $\hat{Z}$-strings) and **fluxes** $m$ ($B_p = -1$, created by $\hat{X}$-strings).
- The $\hat{Z}$-string and $\hat{X}$-loop anticommute when they cross ({eq}`eq-string-anticommute`), giving an $e$-$m$ braiding phase of $\pi$: this identifies $e$ as unit charge and $m$ as $\pi$-flux.
- The composite $\varepsilon = e \times m$ is a **fermion** — fermionic statistics emerges from bosonic qubits.
- The four anyon sectors $\{1, e, m, \varepsilon\}$ form a $\mathbb{Z}_2 \times \mathbb{Z}_2$ fusion algebra; all mutual statistics follow from the charge-flux composite picture.
- The toric code is a **topological quantum error-correcting code**: logical errors require system-spanning strings, providing exponential protection.
