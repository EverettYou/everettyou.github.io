## Lecture Notes

### From Qubits to Particles

One of the most remarkable facts in modern physics is that **particles can emerge from qubits**. The toric code, introduced by Alexei Kitaev in 2003, is the simplest exactly solvable model demonstrating this. Starting from a lattice of spin-1/2 qubits with purely local interactions, the model produces emergent excitations that behave as anyons — including objects that are effectively **bosons** and **fermions**, even though the microscopic degrees of freedom are just qubits on lattice edges.

:::{admonition} Toric Code Hamiltonian
:class: important

Place a qubit (spin-1/2) on every **edge** of a square lattice on a torus. Define two types of operators:

- **Star operator** $A_v = \prod_{e \ni v} \hat{\sigma}^x_e$ — product of $\hat{\sigma}^x$ on all edges touching vertex $v$
- **Plaquette operator** $B_p = \prod_{e \in p} \hat{\sigma}^z_e$ — product of $\hat{\sigma}^z$ on all edges around plaquette $p$

The Hamiltonian is:

$$H = -\sum_v A_v - \sum_p B_p$$

All terms commute: $[A_v, A_{v'}] = [B_p, B_{p'}] = [A_v, B_p] = 0$. The ground state satisfies $A_v = +1$ and $B_p = +1$ for all $v$ and $p$.
:::

### Two Types of Excitations

Excitations are created by violating either the star or plaquette constraint:

:::{admonition} Charges and Fluxes
:class: important

| Excitation | Violated constraint | Created by | Name |
|------------|-------------------|------------|------|
| **Charge** ($e$) | $A_v = -1$ at vertex $v$ | String of $\hat{\sigma}^z$ operators | Electric charge |
| **Flux** ($m$) | $B_p = -1$ at plaquette $p$ | String of $\hat{\sigma}^x$ operators | Magnetic flux |

- Charges are created in **pairs** at the endpoints of a $\hat{\sigma}^z$-string
- Fluxes are created in **pairs** at the endpoints of a $\hat{\sigma}^x$-string
- Both charges and fluxes are **bosons** individually: exchanging two charges (or two fluxes) gives phase $+1$
:::

### Emergent Fermion from Charge-Flux Binding

Here is the key insight: while charges and fluxes are individually bosons, **binding a charge to a flux produces a fermion**.

:::{admonition} The Composite Fermion $\varepsilon = e \times m$
:class: important

The composite particle $\varepsilon$ (a charge-flux bound state) has exchange statistics:

$$\theta_\varepsilon = \pi \quad \Longrightarrow \quad \text{fermion!}$$

This is understood through the **charge-flux picture**: when one $\varepsilon$ particle goes around another, the charge part of the first sees the flux part of the second (and vice versa), picking up an Aharonov-Bohm phase of $\pi$ from each contribution, totaling $2\pi$ for a full loop, hence $\pi$ for an exchange.
:::

:::{admonition} Mutual Statistics
:class: note

The charge and flux also have nontrivial **mutual statistics**: taking a charge $e$ around a flux $m$ (a full loop) gives a phase of $-1$:

$$\theta_{em} = \pi$$

This is again the Aharonov-Bohm effect. It is this mutual $\pi$ phase that makes the composite $\varepsilon = e \times m$ a fermion.
:::

### Summary of Emergent Particles

| Particle | Composition | Self-statistics | Physical analogy |
|----------|------------|----------------|-----------------|
| $e$ (charge) | Star violation | Boson ($\theta = 0$) | Electric charge |
| $m$ (flux) | Plaquette violation | Boson ($\theta = 0$) | Magnetic flux |
| $\varepsilon = e \times m$ | Bound state | **Fermion** ($\theta = \pi$) | Electron-like |
| $1$ (vacuum) | No excitation | — | Ground state |

The remarkable conclusion: **fermions are not fundamental** in the toric code. They emerge as composite objects from a system of qubits that has no fermions in its microscopic description.

:::{admonition} Toric Code as a Quantum Error-Correcting Code
:class: important

The toric code is not just a physics toy model — it is one of the most important quantum error-correcting codes:

- **Logical qubits** are encoded in the degenerate ground-state space (4-fold degenerate on a torus)
- **Errors** (bit flips $\hat{\sigma}^x$ or phase flips $\hat{\sigma}^z$) create pairs of excitations
- **Error detection**: measure all star and plaquette operators to locate excitation pairs
- **Error correction**: connect the excitation pairs with minimal strings to annihilate them
- **Topological protection**: a logical error requires a string of errors spanning the entire torus — exponentially unlikely for small error rates
:::

The toric code demonstrates the principle of **topological quantum error correction**: information is stored nonlocally in the topology of the ground state, making it immune to any local perturbation. This idea is the foundation of modern approaches to building fault-tolerant quantum computers using **surface codes** (the planar version of the toric code), which are currently the leading architecture for experimental quantum error correction.

:::{admonition} Discussion: Particles from Qubits
:class: dropdown warning

The toric code shows that both bosons and fermions can emerge from a system of qubits. Does this mean that the electrons and photons in our universe could themselves be emergent excitations of some deeper "qubit" substrate? What would it take to get emergent particles with the full structure of the Standard Model (chirality, gauge symmetry, gravity)? This is an active area of research.
:::
