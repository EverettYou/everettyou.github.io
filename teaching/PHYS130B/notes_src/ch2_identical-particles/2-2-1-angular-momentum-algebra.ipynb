## Lecture Notes

### Overview

In Chapter 1, we studied a single qubit â€” the spin-1/2 system with two eigenstates. This raises a natural question: **what about higher spin?** What constrains the possible values of angular momentum, and how do they arise?

The answer is remarkable: angular momentum quantization follows entirely from **commutation relations**, without solving any differential equation. All forms of angular momentum â€” orbital ($\boldsymbol{L} = \boldsymbol{r} \times \boldsymbol{p}$), spin ($\boldsymbol{S}$), and total ($\boldsymbol{J} = \boldsymbol{L} + \boldsymbol{S}$) â€” satisfy the same algebra. This algebra is the **Lie algebra of SU(2)**, the group of rotations we met in [Â§1.3.3](../ch1_qubit/1-3-3-heisenberg-picture). The physical content is simple: rotations in 3D do not commute, and angular momentum operators are their generators.

### Angular Momentum Commutation Relations

:::{admonition} Defining Relation
:class: important

The three components of any angular momentum operator $\hat{\boldsymbol{J}} = (\hat{J}_x, \hat{J}_y, \hat{J}_z)$ satisfy:

$$
[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar \epsilon_{ijk} \hat{J}_k
$$ (eq-am-commutation)

Explicitly:

$$
[\hat{J}_x, \hat{J}_y] = \mathrm{i}\hbar \hat{J}_z, \quad [\hat{J}_y, \hat{J}_z] = \mathrm{i}\hbar \hat{J}_x, \quad [\hat{J}_z, \hat{J}_x] = \mathrm{i}\hbar \hat{J}_y
$$

These are **cyclic** in $(x, y, z)$. The Levi-Civita symbol $\epsilon_{ijk}$ is $+1$ for even permutations of $(1,2,3)$, $-1$ for odd, and $0$ if any index repeats.
:::

For orbital angular momentum, these follow from the canonical commutation relation $[\hat{r}_i, \hat{p}_j] = \mathrm{i}\hbar \delta_{ij}$. For spin, they are **postulated** â€” the requirement is that angular momentum generates rotations, so it must satisfy the Lie algebra of SO(3).

:::{admonition} Derivation: Orbital Angular Momentum Commutators
:class: dropdown information

Starting from $\hat{L}_i = \epsilon_{imn} \hat{r}_m \hat{p}_n$ and the canonical commutation relation $[\hat{r}_m, \hat{p}_q] = \mathrm{i}\hbar \delta_{mq}$, we compute $[\hat{L}_x, \hat{L}_y]$ explicitly. With $\hat{L}_x = \hat{y}\hat{p}_z - \hat{z}\hat{p}_y$ and $\hat{L}_y = \hat{z}\hat{p}_x - \hat{x}\hat{p}_z$:

$$
[\hat{L}_x, \hat{L}_y] = [\hat{y}\hat{p}_z - \hat{z}\hat{p}_y,\; \hat{z}\hat{p}_x - \hat{x}\hat{p}_z]
$$

Expanding the four commutator terms using $[\hat{A}\hat{B}, \hat{C}\hat{D}] = \hat{A}[\hat{B}, \hat{C}]\hat{D} + \hat{C}[\hat{A}, \hat{D}]\hat{B} + \ldots$ and noting that operators acting on different coordinates commute:

$$
= [\hat{y}\hat{p}_z, \hat{z}\hat{p}_x] - [\hat{y}\hat{p}_z, \hat{x}\hat{p}_z] - [\hat{z}\hat{p}_y, \hat{z}\hat{p}_x] + [\hat{z}\hat{p}_y, \hat{x}\hat{p}_z]
$$

The middle two terms vanish (no conjugate pairs share a coordinate). The surviving terms give:

$$
= \hat{y}[\hat{p}_z, \hat{z}]\hat{p}_x + \hat{x}[\hat{z}, \hat{p}_z]\hat{p}_y = \hat{y}(-\mathrm{i}\hbar)\hat{p}_x + \hat{x}(\mathrm{i}\hbar)\hat{p}_y = \mathrm{i}\hbar(\hat{x}\hat{p}_y - \hat{y}\hat{p}_x) = \mathrm{i}\hbar \hat{L}_z
$$

The other two relations follow by cyclic permutation $(x \to y \to z \to x)$. $\checkmark$

**General form.** In index notation with $\hat{L}_i = \epsilon_{imn}\hat{r}_m\hat{p}_n$, the product of Levi-Civita symbols satisfies the contraction identity $\epsilon_{imn}\epsilon_{jpq} = \delta_{ij}(\delta_{mp}\delta_{nq} - \delta_{mq}\delta_{np}) + \text{cyclic}$, which reduces the general commutator to $[\hat{L}_i, \hat{L}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{L}_k$.
:::

:::{admonition} Connection to Rotations
:class: note

The commutation relations encode the geometry of 3D rotations: rotating first about $x$ then about $y$ differs from $y$ then $x$. Mathematically, $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar \epsilon_{ijk} \hat{J}_k$ is the **Lie algebra $\mathfrak{su}(2)$** â€” the same algebra that appeared in [Â§1.3.3](../ch1_qubit/1-3-3-heisenberg-picture) as the algebra of SU(2) generators. The angular momentum operators $\hat{J}_i$ generate rotations via $\hat{U}(\boldsymbol{n}, \theta) = \mathrm{e}^{-\mathrm{i}\theta \boldsymbol{n} \cdot \hat{\boldsymbol{J}}/\hbar}$.
:::

### Casimir Operator $\hat{J}^2$

:::{admonition} Casimir Operator
:class: important

The **total angular momentum squared**:

$$
\hat{J}^2 = \hat{J}_x^2 + \hat{J}_y^2 + \hat{J}_z^2
$$ (eq-casimir)

commutes with every component:

$$
[\hat{J}^2, \hat{J}_i] = 0 \quad \text{for } i = x, y, z
$$ (eq-casimir-commutes)
:::

Because $[\hat{J}^2, \hat{J}_z] = 0$, we can find **simultaneous eigenstates** of both operators, labeled $\vert j, m\rangle$.

:::{admonition} Derivation: $[\hat{J}^2, \hat{J}_z] = 0$
:class: dropdown information

$$
[\hat{J}^2, \hat{J}_z] = [\hat{J}_x^2, \hat{J}_z] + [\hat{J}_y^2, \hat{J}_z]
$$

Using $[\hat{A}^2, \hat{B}] = \hat{A}[\hat{A}, \hat{B}] + [\hat{A}, \hat{B}]\hat{A}$:

$$
[\hat{J}_x^2, \hat{J}_z] = -\mathrm{i}\hbar(\hat{J}_x \hat{J}_y + \hat{J}_y \hat{J}_x), \quad [\hat{J}_y^2, \hat{J}_z] = +\mathrm{i}\hbar(\hat{J}_y \hat{J}_x + \hat{J}_x \hat{J}_y)
$$

These cancel exactly. $\checkmark$
:::

### Ladder Operators

:::{admonition} Ladder Operators
:class: important

The **raising** and **lowering** operators:

$$
\hat{J}_+ = \hat{J}_x + \mathrm{i}\hat{J}_y, \quad \hat{J}_- = \hat{J}_x - \mathrm{i}\hat{J}_y
$$ (eq-ladder-def)

satisfy:

$$
[\hat{J}_z, \hat{J}_\pm] = \pm\hbar \hat{J}_\pm, \quad [\hat{J}_+, \hat{J}_-] = 2\hbar \hat{J}_z, \quad [\hat{J}^2, \hat{J}_\pm] = 0
$$
:::

:::{admonition} Derivation: Ladder Operator Commutation Relations
:class: dropdown information

**$[\hat{J}_z, \hat{J}_\pm] = \pm\hbar \hat{J}_\pm$:** Using $\hat{J}_+ = \hat{J}_x + \mathrm{i}\hat{J}_y$:

$$
[\hat{J}_z, \hat{J}_+] = [\hat{J}_z, \hat{J}_x] + \mathrm{i}[\hat{J}_z, \hat{J}_y] = \mathrm{i}\hbar\hat{J}_y + \mathrm{i}(-\mathrm{i}\hbar\hat{J}_x) = \hbar(\hat{J}_x + \mathrm{i}\hat{J}_y) = +\hbar\hat{J}_+
$$

Similarly $[\hat{J}_z, \hat{J}_-] = -\hbar\hat{J}_-$. This means $\hat{J}_\pm$ shifts the $\hat{J}_z$ eigenvalue by $\pm\hbar$.

**$[\hat{J}_+, \hat{J}_-] = 2\hbar\hat{J}_z$:**

$$
[\hat{J}_+, \hat{J}_-] = [\hat{J}_x + \mathrm{i}\hat{J}_y,\, \hat{J}_x - \mathrm{i}\hat{J}_y] = -\mathrm{i}[\hat{J}_x, \hat{J}_y] + \mathrm{i}[\hat{J}_y, \hat{J}_x] = -2\mathrm{i}[\hat{J}_x, \hat{J}_y] = -2\mathrm{i}(\mathrm{i}\hbar\hat{J}_z) = 2\hbar\hat{J}_z
$$

**$[\hat{J}^2, \hat{J}_\pm] = 0$:** Follows from $[\hat{J}^2, \hat{J}_x] = [\hat{J}^2, \hat{J}_y] = 0$, since $\hat{J}_\pm$ is a linear combination of $\hat{J}_x$ and $\hat{J}_y$. $\checkmark$
:::

The key property: $\hat{J}_+$ raises $m$ by one, $\hat{J}_-$ lowers it by one.

:::{admonition} Ladder Operator Action
:class: important

$$
\hat{J}_+ \vert j, m\rangle = \hbar\sqrt{(j-m)(j+m+1)}\;\vert j, m+1\rangle
$$ (eq-raising)

$$
\hat{J}_- \vert j, m\rangle = \hbar\sqrt{(j+m)(j-m+1)}\;\vert j, m-1\rangle
$$ (eq-lowering)

The coefficient vanishes at the boundaries: $\hat{J}_+ \vert j, j\rangle = 0$ and $\hat{J}_- \vert j, -j\rangle = 0$.
:::

:::{admonition} Derivation: Normalization Coefficients
:class: dropdown information

Let $\vert\psi\rangle = \hat{J}_+ \vert j, m\rangle$. From $[\hat{J}_z, \hat{J}_+] = \hbar \hat{J}_+$:

$$
\hat{J}_z \vert\psi\rangle = \hbar(m+1)\vert\psi\rangle
$$

From $[\hat{J}^2, \hat{J}_+] = 0$: $\hat{J}^2 \vert\psi\rangle = \hbar^2 j(j+1)\vert\psi\rangle$. So $\vert\psi\rangle \propto \vert j, m+1\rangle$.

For the norm, first derive the identity $\hat{J}_- \hat{J}_+ = \hat{J}^2 - \hat{J}_z^2 - \hbar\hat{J}_z$. Writing $\hat{J}_x = \frac{1}{2}(\hat{J}_+ + \hat{J}_-)$ and $\hat{J}_y = \frac{1}{2\mathrm{i}}(\hat{J}_+ - \hat{J}_-)$:

$$
\hat{J}_x^2 + \hat{J}_y^2 = \frac{1}{2}(\hat{J}_+\hat{J}_- + \hat{J}_-\hat{J}_+) = \hat{J}^2 - \hat{J}_z^2
$$

Using $[\hat{J}_+, \hat{J}_-] = 2\hbar\hat{J}_z$ gives $\hat{J}_+\hat{J}_- = \hat{J}_-\hat{J}_+ + 2\hbar\hat{J}_z$, so:

$$
\hat{J}_-\hat{J}_+ = \hat{J}^2 - \hat{J}_z^2 - \hbar\hat{J}_z, \qquad \hat{J}_+\hat{J}_- = \hat{J}^2 - \hat{J}_z^2 + \hbar\hat{J}_z
$$

Then:

$$
\|\hat{J}_+ \vert j, m\rangle\|^2 = \langle j,m\vert \hat{J}_- \hat{J}_+ \vert j,m\rangle = \hbar^2[j(j+1) - m(m+1)] = \hbar^2(j-m)(j+m+1)
$$

Taking the square root gives the coefficient. $\checkmark$
:::

### Angular Momentum Quantization

The quantization of $j$ and $m$ follows purely from the algebra â€” no differential equation needed.

:::{admonition} Quantization Rules
:class: important

$$
j \in \{0,\; \tfrac{1}{2},\; 1,\; \tfrac{3}{2},\; 2,\; \ldots\}, \qquad m \in \{-j,\; -j+1,\; \ldots,\; j-1,\; j\}
$$ (eq-quantization)

For each $j$, there are $2j+1$ states. The eigenvalues are:

$$
\hat{J}^2 \vert j,m\rangle = \hbar^2 j(j+1)\vert j,m\rangle, \qquad \hat{J}_z \vert j,m\rangle = \hbar m \vert j,m\rangle
$$
:::

:::{admonition} Derivation: Why $j$ Is Quantized
:class: dropdown information

For fixed $j$, the ladder operators generate a chain of states with $m$ values differing by integers. Since $\langle \hat{J}_x^2 + \hat{J}_y^2\rangle \geq 0$ implies $m^2 \leq j(j+1)$, the chain must terminate: there exist $m_{\max}$ and $m_{\min}$ with $\hat{J}_+ \vert j, m_{\max}\rangle = 0$ and $\hat{J}_- \vert j, m_{\min}\rangle = 0$.

From the norm formula: $m_{\max} = j$ and $m_{\min} = -j$. Since $m$ changes in integer steps, we need $m_{\max} - m_{\min} = 2j \in \mathbb{Z}$, so $j$ is a non-negative integer or half-integer. $\checkmark$
:::

**Integer vs. half-integer:** Orbital angular momentum ($\hat{\boldsymbol{L}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{p}}$) only admits integer $\ell$ because the wavefunctions $Y_\ell^m(\theta,\phi)$ must be single-valued on the sphere. Spin has no spatial wavefunction, so half-integer $j$ is allowed. The **spin-statistics theorem** (proved in QFT) links this to particle statistics: integer spin $\to$ bosons, half-integer spin $\to$ fermions.

:::{admonition} Discussion
:class: dropdown tip

Angular momentum quantization follows from algebra alone â€” no differential equation, no specific Hamiltonian. Does this mean any operator satisfying $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$ must have quantized eigenvalues? Could there be physical systems where the algebra is deformed (e.g., $q$-deformed angular momentum)? What would change?
:::

### Matrix Representations

For a given $j$, the operators act on the $(2j+1)$-dimensional space spanned by $\{\vert j, m\rangle : m = -j, \ldots, j\}$.

:::{admonition} Spin-1/2: Pauli Matrices
:class: important

For $j = 1/2$, the two basis states are $\vert\uparrow\rangle = \vert\tfrac{1}{2}, +\tfrac{1}{2}\rangle$ and $\vert\downarrow\rangle = \vert\tfrac{1}{2}, -\tfrac{1}{2}\rangle$. The angular momentum operators are $\hat{J}_i = \frac{\hbar}{2}\hat{\sigma}^i$:

$$
\hat{J}_z = \frac{\hbar}{2}\begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}, \quad \hat{J}_x = \frac{\hbar}{2}\begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}, \quad \hat{J}_y = \frac{\hbar}{2}\begin{pmatrix} 0 & -\mathrm{i} \\ \mathrm{i} & 0 \end{pmatrix}
$$

These are the Pauli matrices from [Â§1.1.3](../ch1_qubit/1-1-3-hermitian-operators), now understood as the $j = 1/2$ representation of angular momentum.
:::

:::{admonition} Spin-1: 3Ã—3 Matrices
:class: important

For $j = 1$, three basis states $\vert 1, +1\rangle$, $\vert 1, 0\rangle$, $\vert 1, -1\rangle$:

$$
\hat{J}_z = \hbar\begin{pmatrix} 1 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & -1 \end{pmatrix}, \quad \hat{J}_x = \frac{\hbar}{\sqrt{2}}\begin{pmatrix} 0 & 1 & 0 \\ 1 & 0 & 1 \\ 0 & 1 & 0 \end{pmatrix}, \quad \hat{J}_y = \frac{\hbar}{\sqrt{2}}\begin{pmatrix} 0 & -\mathrm{i} & 0 \\ \mathrm{i} & 0 & -\mathrm{i} \\ 0 & \mathrm{i} & 0 \end{pmatrix}
$$

The factors of $\sqrt{2}$ come from the ladder operator formula with $j = 1$.
:::

**General pattern:** For spin-$j$, the operators are $(2j+1) \times (2j+1)$ matrices. Each representation is **irreducible** â€” it cannot be block-diagonalized while preserving the algebra.

### Summary

- All angular momenta satisfy the **same commutation relations** $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$ â€” the Lie algebra $\mathfrak{su}(2)$.
- The **Casimir operator** $\hat{J}^2$ commutes with all components, enabling simultaneous eigenstates $\vert j, m\rangle$.
- **Quantization** ($j = 0, \frac{1}{2}, 1, \ldots$; $m = -j, \ldots, j$) follows from algebra alone.
- **Ladder operators** $\hat{J}_\pm$ shift $m$ by $\pm 1$ with known normalization.
- **Representations**: spin-1/2 recovers Pauli matrices; spin-1 gives $3\times 3$ matrices; general $j$ gives $(2j+1)$-dimensional irreducible representations.
