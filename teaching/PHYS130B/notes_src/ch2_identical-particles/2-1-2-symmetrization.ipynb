## Lecture Notes

Identical particles in quantum mechanics are not merely indistinguishable; they obey profound symmetry constraints: fermions must have antisymmetric wave functions, while bosons must have symmetric ones. This section shows how this fundamental distinction arises from relativity and particle statistics, and how it shapes the structure of matter.
### The Problem of Labeling

In classical mechanics, particles are distinguishable: we can label them as "particle 1," "particle 2," etc., and track each one's trajectory.

In quantum mechanics, particles of the same type (e.g., two electrons) are **absolutely identical**. They have:
- The same mass
- The same charge
- The same spin
- **No inherent labels or identity tags**

This has profound consequences: If we prepare two identical particles in a given state and then ask "which particle is where?," the question itself is **meaningless**. Nature provides no answer.

### Physical Indistinguishability

The physical requirement is that swapping any two identical particles leaves the probability density unchanged:

$$

|\psi(\boldsymbol{r}_1, \boldsymbol{r}_2, \ldots, \boldsymbol{r}_N)|^2 = |\psi(\boldsymbol{r}_{\sigma(1)}, \boldsymbol{r}_{\sigma(2)}, \ldots, \boldsymbol{r}_{\sigma(N)})|^2

$$ (eq-indistinguishability)

for any permutation $\sigma$.

:::{admonition} Pauli exclusion
:class: important
Two identical fermions cannot occupy the same quantum state. This follows directly from the antisymmetry of the wave function: if two particles share a state, the Slater determinant vanishes identically.
:::

- Key takeaway 3
:::{important} Permutation Operators
:class: important

Permutation operators Pᵢⱼ swap particles i and j. For identical particles, physical states must be eigenstates of permutation operators, with eigenvalues ±1 for fermions and bosons respectively.
:::

:::{note} Slater Determinant Construction
:class: note

The Slater determinant is a compact way to encode fermionic antisymmetry and automatically enforce Pauli exclusion. For N particles in orbitals {φₖ}, it is det([φ₁(r₁) φ₂(r₁) ... ; φ₁(r₂) φ₂(r₂) ... ; ...]).
:::

:::{warning} Pauli Exclusion Principle
:class: warning

No two fermions can occupy the same state. Mathematically, the fermionic wavefunction vanishes if any two particles occupy identical single-particle states. This is not an ad-hoc rule but a direct consequence of antisymmetry.
:::

:::{caution} Boson Accumulation
:class: caution

Unlike fermions, multiple bosons can occupy the same single-particle state. This is why Bose-Einstein condensation is possible—at low temperatures, many identical bosons can settle into the ground state.
:::


## The Permutation Operator and Its Eigenvalues

### Definition of the Permutation Operator

The **permutation operator** $P^\pi$ corresponding to permutation $\pi$ acts on a many-body state by relabeling spatial arguments:

$$

P^\pi |\psi(\boldsymbol{r}_1, \boldsymbol{r}_2, \ldots, \boldsymbol{r}_N)\rangle = |\psi(\boldsymbol{r}_{\pi(1)}, \boldsymbol{r}_{\pi(2)}, \ldots, \boldsymbol{r}_{\pi(N)})\rangle

$$ (eq-permutation-operator)

**Example**: The exchange of particles 1 and 2:

$$

P_{12} \psi(\boldsymbol{r}_1, \boldsymbol{r}_2, \boldsymbol{r}_3, \ldots) = \psi(\boldsymbol{r}_2, \boldsymbol{r}_1, \boldsymbol{r}_3, \ldots)

$$

### Physical Requirement: Phase Eigenvalue

Since probability density must be unchanged under permutation:

$$

|(P^\pi \psi)(\boldsymbol{r}_1, \ldots)|^2 = |\psi(\boldsymbol{r}_1, \ldots)|^2

$$

this means $P^\pi$ can only change $\psi$ by an **overall phase factor**:

$$

P^\pi \psi = e^{i\theta_\pi} \psi

$$ (eq-permutation-eigenvalue)

::{admonition} Derivation: Constraint on Eigenvalues from Double Exchange
:class: dropdown note

For a two-particle exchange applied twice:

$$

P_{12}^2 \psi = \psi \quad \Rightarrow \quad e^{2i\theta} = 1 \quad \Rightarrow \quad e^{i\theta} = \pm 1

$$

**Only two possibilities** exist for identical particles:

$$

P_{12} \psi = +\psi \quad \text{(symmetric, bosons)}

$$

$$

P_{12} \psi = -\psi \quad \text{(antisymmetric, fermions)}

$$ (eq-boson-fermion-phases)

:::

## Symmetric and Antisymmetric States

### Bosonic (Symmetric) States

A **symmetric** state is invariant under all particle permutations:

$$

\psi_B(\boldsymbol{r}_1, \boldsymbol{r}_2, \ldots, \boldsymbol{r}_N) = \psi_B(\boldsymbol{r}_{\sigma(1)}, \boldsymbol{r}_{\sigma(2)}, \ldots, \boldsymbol{r}_{\sigma(N)})

$$ (eq-symmetric-state)

for **all** permutations $\sigma$.

**Construction**: Given $N$ distinct single-particle wavefunctions $\phi_1(\boldsymbol{r}), \phi_2(\boldsymbol{r}), \ldots, \phi_N(\boldsymbol{r})$, the **symmetrized product** is:

$$

\psi_B(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_N) = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \phi_{\pi(1)}(\boldsymbol{r}_1) \phi_{\pi(2)}(\boldsymbol{r}_2) \cdots \phi_{\pi(N)}(\boldsymbol{r}_N)

$$ (eq-boson-state-construction)

where the sum runs over all $N!$ permutations of $\{1, 2, \ldots, N\}$.

**Key properties**:
- Multiple bosons can occupy the **same** single-particle state.
- Bose-Einstein condensation: in equilibrium, a macroscopic number can occupy the ground state.
- Examples: photons, $^4$He nuclei, phonons.

### Fermionic (Antisymmetric) States

An **antisymmetric** state changes sign under all transpositions (two-particle exchanges):

$$

\psi_F(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_i, \ldots, \boldsymbol{r}_j, \ldots) = -\psi_F(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_j, \ldots, \boldsymbol{r}_i, \ldots)

$$ (eq-antisymmetric-state)

More generally, for a permutation $\pi$ with **signature** $\text{sgn}(\pi) = \pm 1$ (even or odd number of transpositions):

$$

P^\pi \psi_F = \text{sgn}(\pi) \cdot \psi_F

$$

**Construction**: The **Slater determinant** (see next subsection) automatically encodes this antisymmetry.

## The Pauli Exclusion Principle

### Derivation from Antisymmetry

Consider two identical fermions with an **antisymmetric** wavefunction. Set both particles at the **same position** $\boldsymbol{r}_i = \boldsymbol{r}_j = \boldsymbol{r}$:

$$

\psi_F(\boldsymbol{r}, \boldsymbol{r}, \boldsymbol{r}_3, \ldots) = -\psi_F(\boldsymbol{r}, \boldsymbol{r}, \boldsymbol{r}_3, \ldots)

$$

The only function satisfying $f = -f$ is $f = 0$. Therefore:

$$

\psi_F(\boldsymbol{r}, \boldsymbol{r}, \boldsymbol{r}_3, \ldots) = 0

$$ (eq-pauli-exclusion-1)

**In the context of single-particle states**: If two fermions occupy identical single-particle states (same spatial wavefunction, same spin), the many-body wavefunction vanishes.

$$

\boxed{\text{No two identical fermions can occupy the same single-particle quantum state.}}

$$ (eq-pauli-exclusion-principle)

### Occupation Number Constraints

In terms of occupation numbers $n_\alpha$ (number of fermions in single-particle state $\alpha$):

$$

n_\alpha \in \{0, 1\} \quad \text{(fermions only)}

$$

For bosons, there is no such restriction:

$$

n_\alpha \in \{0, 1, 2, 3, \ldots\} \quad \text{(bosons)}

$$

### Consequences

The Pauli exclusion principle has profound physical consequences:

1. **Finite density**: Fermionic matter has a maximum density; particles cannot be compressed beyond a certain limit.

2. **Fermi surface**: At zero temperature, fermions fill states up to a maximum energy (Fermi energy), creating a sharp boundary in momentum space.

3. **Degeneracy pressure**: The exclusion principle creates an effective repulsion that stabilizes compact objects (white dwarfs, neutron stars).

4. **Chemistry**: The shell structure of atoms and chemical bonding arise from electrons filling discrete orbitals subject to exclusion.

5. **Ferromagnetism**: Spin alignment in magnetic materials is partly enforced by exclusion combined with exchange interactions.

## Slater Determinants

### Definition

The **Slater determinant** is a convenient way to construct an **antisymmetric** many-particle wavefunction from $N$ single-particle wavefunctions $\phi_1, \phi_2, \ldots, \phi_N$:

$$

\Psi(\boldsymbol{r}_1, \boldsymbol{r}_2, \ldots, \boldsymbol{r}_N) = \frac{1}{\sqrt{N!}} \begin{vmatrix}
\phi_1(\boldsymbol{r}_1) & \phi_2(\boldsymbol{r}_1) & \cdots & \phi_N(\boldsymbol{r}_1) \
\phi_1(\boldsymbol{r}_2) & \phi_2(\boldsymbol{r}_2) & \cdots & \phi_N(\boldsymbol{r}_2) \
\vdots & \vdots & \ddots & \vdots \
\phi_1(\boldsymbol{r}_N) & \phi_2(\boldsymbol{r}_N) & \cdots & \phi_N(\boldsymbol{r}_N)
\end{vmatrix}

$$ (eq-slater-determinant)

The factor $1/\sqrt{N!}$ ensures normalization (assuming $\phi_i$ are orthonormal).

### Properties

**1. Automatic Antisymmetry**

Swapping two rows of a determinant changes its sign:

$$

\Psi(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_i, \ldots, \boldsymbol{r}_j, \ldots) = -\Psi(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_j, \ldots, \boldsymbol{r}_i, \ldots)

$$

This is a **fundamental property of determinants**, so antisymmetry is built in automatically.

**2. Automatic Pauli Exclusion**

If two orbitals are identical ($\phi_i = \phi_j$), then two columns of the determinant are identical, so:

$$

\Psi = 0

$$

This automatically encodes the Pauli exclusion principle.

**3. Expansion Form**

Expanding the Slater determinant explicitly:

$$

\Psi(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_N) = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \text{sgn}(\pi) \, \phi_{\pi(1)}(\boldsymbol{r}_1) \phi_{\pi(2)}(\boldsymbol{r}_2) \cdots \phi_{\pi(N)}(\boldsymbol{r}_N)

$$ (eq-slater-expansion)

The sum over permutations, weighted by their signature, creates antisymmetry.

## Permanents: Symmetric Analog of Determinants

For **bosonic** states, we might wish to construct a symmetric wavefunction. The analog of the Slater determinant is the **permanent**:

$$

\Psi_B(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_N) = \frac{1}{\sqrt{N!}} \sum_{\pi \in S_N} \phi_{\pi(1)}(\boldsymbol{r}_1) \phi_{\pi(2)}(\boldsymbol{r}_2) \cdots \phi_{\pi(N)}(\boldsymbol{r}_N)

$$ (eq-permanent)

Note: The permanent is like a determinant but **all signs are positive** (no $\text{sgn}(\pi)$ factor).

**Symmetry property**:

$$

\Psi_B(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_i, \ldots, \boldsymbol{r}_j, \ldots) = +\Psi_B(\boldsymbol{r}_1, \ldots, \boldsymbol{r}_j, \ldots, \boldsymbol{r}_i, \ldots)

$$

The permanent is less commonly used in calculations (it's harder to compute), but it appears in certain contexts like photon and boson condensate wavefunctions.

:::{admonition} Example: Two Identical Fermions
:class: example


### Setup

Consider two identical fermions in single-particle states $|\phi_1\rangle$ and $|\phi_2\rangle$.

### The Slater Determinant

$$

\Psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = \frac{1}{\sqrt{2}} \begin{vmatrix}
\phi_1(\boldsymbol{r}_1) & \phi_2(\boldsymbol{r}_1) \
\phi_1(\boldsymbol{r}_2) & \phi_2(\boldsymbol{r}_2)
\end{vmatrix}

$$

$$

= \frac{1}{\sqrt{2}} \left[ \phi_1(\boldsymbol{r}_1) \phi_2(\boldsymbol{r}_2) - \phi_1(\boldsymbol{r}_2) \phi_2(\boldsymbol{r}_1) \right]

$$ (eq-two-fermion-slater)

### Check Antisymmetry

Exchange $\boldsymbol{r}_1 \leftrightarrow \boldsymbol{r}_2$:

$$

\Psi(\boldsymbol{r}_2, \boldsymbol{r}_1) = \frac{1}{\sqrt{2}} \left[ \phi_1(\boldsymbol{r}_2) \phi_2(\boldsymbol{r}_1) - \phi_1(\boldsymbol{r}_1) \phi_2(\boldsymbol{r}_2) \right] = -\Psi(\boldsymbol{r}_1, \boldsymbol{r}_2)

$$

✓ Antisymmetry is satisfied.

### Check Pauli Exclusion

If $\phi_1 = \phi_2$:

$$

\Psi(\boldsymbol{r}_1, \boldsymbol{r}_2) = \frac{1}{\sqrt{2}} \left[ \phi(\boldsymbol{r}_1) \phi(\boldsymbol{r}_2) - \phi(\boldsymbol{r}_2) \phi(\boldsymbol{r}_1) \right] = 0

$$

✓ Both fermions cannot be in the same state.

### Physical Interpretation

The Slater determinant represents a state where:
- Particle 1 is equally likely to be in $\phi_1$ or $\phi_2$.
- Particle 2 is in the complementary state.
- The particles are **completely entangled** in their orbital state.
:::

## Symmetry, Spin, and Total Wavefunction

### The Full Wavefunction: Spatial and Spin

For fermions with both spatial and spin degrees of freedom, the **total wavefunction** is:

$$

\Psi_{\text{total}}(\boldsymbol{r}_1, s_1; \boldsymbol{r}_2, s_2; \ldots) = \Psi_\text{spatial}(\boldsymbol{r}_1, \boldsymbol{r}_2, \ldots) \times \chi_\text{spin}(s_1, s_2, \ldots)

$$ (eq-total-fermion-wavefunction)

**Fundamental constraint**: The total wavefunction must be **antisymmetric** under all particle exchanges.

$$

\Psi_\text{total} = \Psi_\text{spatial} \times \chi_\text{spin} \quad \text{is antisymmetric}

$$

This means:
- If $\Psi_\text{spatial}$ is **symmetric**, then $\chi_\text{spin}$ must be **antisymmetric** (singlet).
- If $\Psi_\text{spatial}$ is **antisymmetric**, then $\chi_\text{spin}$ must be **symmetric** (triplet).

:::{admonition} Example: Two Electrons (Spin-1/2)
:class: example


**Case 1**: Both in the same spatial orbital (ground state of helium), so $\Psi_\text{spatial}$ is **symmetric**.

Then $\chi_\text{spin}$ must be a **singlet** (antisymmetric):

$$

\chi_\text{singlet} = \frac{1}{\sqrt{2}}(|\uparrow\downarrow\rangle - |\downarrow\uparrow\rangle)

$$

The electrons must have **opposite spins**. This is the **spin-paired** configuration.

**Case 2**: In different spatial orbitals (excited state), so $\Psi_\text{spatial}$ is **antisymmetric**.

Then $\chi_\text{spin}$ must be a **triplet** (symmetric):

$$

\chi_\text{triplet} = |\uparrow\uparrow\rangle, \quad \frac{1}{\sqrt{2}}(|\uparrow\downarrow\rangle + |\downarrow\uparrow\rangle), \quad |\downarrow\downarrow\rangle

$$

The electrons can have **parallel spins** and still be in an antisymmetric total state.
:::

## Summary

- **Identical particles** are indistinguishable; swapping two must not change observable probabilities.
- **Permutation operator** can only multiply the wavefunction by $\pm 1$, leading to two classes: symmetric (bosons) and antisymmetric (fermions).
- **Symmetric states** (bosons) allow unlimited occupation per single-particle state.
- **Antisymmetric states** (fermions) enforce the **Pauli exclusion principle**: $n_\alpha \in \{0, 1\}$.
- **Slater determinants** provide a systematic construction for antisymmetric fermionic states.
- **Permanents** are the bosonic analog (symmetric construction).
- **Total wavefunction** must respect the fermionic constraint: if spatial part is symmetric, spin part must be antisymmetric, and vice versa.
