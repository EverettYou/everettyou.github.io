# 2.3 Anyons

## Overview

In 2D systems, particle worldlines can wind around each other without crossing. This topological freedom permits a continuous family of exchange statistics—**anyons**—interpolating between bosons (θ = 0) and fermions (θ = π). Abelian anyons acquire only a phase under exchange; non-abelian anyons acquire matrix-valued statistics. Anyons underpin the fractional quantum Hall effect and enable topological quantum computation.


### Key Concepts

- **Exchange statistics (2D)**: worldlines can wind; statistical phase is $\mathrm{e}^{\mathrm{i}\theta}$ under exchange (braid group generators)
- **Anyons**: $0 \le \theta < 2\pi$ includes all intermediate statistics; Abelian (phase only) and non-abelian (matrix-valued) variants
- **Fractional quantum Hall effect**: filled at $\nu = 1/3$, $1/5$, etc.; quasiparticles are anyons with fractional charge $e/3$, $e/5$
- **Laughlin wavefunction**: $\psi(z_1, \ldots, z_N) \propto \prod_{i<j} (z_i - z_j)^m \exp(-|z_i|^2/4)$; ground state of $\nu = 1/m$
- **Topological quantum gates**: braiding non-abelian anyons encodes quantum information protected from local perturbations
- **Fibonacci anyons**: the simplest non-abelian model; fusion rules $\sigma \times \sigma = 1 + \sigma$ enable universal quantum gates

### Learning Objectives

- Understand why 2D topology permits anyons and how the braid group parametrizes exchange statistics
- Apply Laughlin wavefunction to compute properties of the fractional quantum Hall effect (charge, pair correlation)
- Recognize how non-abelian braiding operators implement quantum gates and why topological protection suppresses errors
- Identify anyon models (Abelian vs. non-abelian) and their relevance to real quantum systems

## Projects

### Project: Anyonic Statistics and Braiding in the Fractional Quantum Hall Effect

**Objective:** Investigate topological order and anyonic braiding statistics in the fractional quantum Hall effect (FQHE). Compute braiding matrices, fusion rules, and simulate the effect of particle exchanges on quantum states.

**Background:** In 2D, indistinguishable particles are not limited to bosons and fermions. **Anyons** have exchange statistics that interpolate continuously, acquiring a topological phase $\theta$ under exchange (Abelian anyons) or even matrix-valued operations (non-abelian anyons). The fractional quantum Hall effect at filling $\nu = 1/3$ exhibits Abelian anyons (charge $e/3$, statistics $\theta = 2\pi/3$). More exotic systems (like $\nu = 5/2$) host non-abelian anyons crucial for topological quantum computation. This project is at the research frontier: understanding topological order and using it for quantum information processing.

**Suggested Approach:**
- **Literature survey**: Study "Anyons in Two Dimensions" (Stern 2008) and "Non-Abelian Anyons and Topological Quantum Computation" (Nayak et al. 2008, Rev. Mod. Phys.). Understand the difference between Abelian (phase) and non-abelian (matrix) statistics.
- **Theory of FQHE**: Review the fractional quantum Hall effect at $\nu = 1/m$ (integer $m$). Understand the Laughlin wavefunction as a Slater determinant of electrons in LLL (lowest Landau level). Show how anyons emerge as quasi-holes (empty sites) in the Laughlin state.
- **Braiding and Fusion**: For $\nu = 1/3$, compute the statistical phase when two anyons are exchanged: $\psi \to \mathrm{e}^{2\pi i / 3} \psi$. Extend to 3+ anyons and derive fusion rules: how do two anyons combine?
- **Computational approach**: Simulate a small FQHE system (6-8 electrons on a torus or disk with periodic boundary conditions). Compute the many-body wavefunction (either Laughlin or lattice model). Track the phase acquired when anyonic excitations are adiabatically exchanged.
- **Benchmarks**: Verify that exchanging two $\nu = 1/3$ anyons twice returns the state to itself (up to the expected phase $4\pi/3$). Compare with analytical predictions.
- **Non-abelian case**: If extended, explore $\nu = 5/2$ and Majorana fermions. Discuss topological quantum computation: how anyonic braiding implements quantum gates.

**Expected Deliverable:** Research report (7-10 pages) with: topological order and anyons overview, Laughlin wavefunction and FQHE, computation of braiding matrices for $\nu = 1/3$, numerical verification of statistical phases, fusion rules, connection to topological quantum computation, and discussion of experimental realization in 2DEGs and recent synthetic platforms.

**Key References:** Stern (2008) "Anyons in two dimensions"; Nayak et al. (2008) "Non-abelian anyons and topological quantum computation"; Das Sarma et al. "Topological order and quantum computation"; papers on FQHE experiments and synthetic anyon platforms (Rydberg atoms, photonics).
