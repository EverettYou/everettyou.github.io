## Lecture Notes

When multiple angular momenta combine—orbital and spin, or spin and spin—their coupling rules determine the allowed energy levels and symmetries of atoms and molecules. This section reveals how the composition of angular momenta generates the rich spectroscopy of the atom and the basis for chemical bonding.

#### Total Angular Momentum

When two particles (or subsystems) each carry angular momentum, the **total angular momentum** is their vector sum:

$$


\boldsymbol{J} = \boldsymbol{J}_1 + \boldsymbol{J}_2

$$ (eq-total-angmom)

The components add:

$$


J_x = J_{1x} + J_{2x}, \quad J_y = J_{1y} + J_{2y}, \quad J_z = J_{1z} + J_{2z}

$$


Both $\boldsymbol{J}_1$ and $\boldsymbol{J}_2$ satisfy angular momentum commutation relations, so their sum automatically satisfies:

$$


[J_i, J_j] = \mathrm{i}\hbar \epsilon_{ijk} J_k

$$

#### The Tensor Product Hilbert Space

If particle 1 has angular momentum quantum numbers $j_1, m_1$ and particle 2 has $j_2, m_2$, the combined system lives in a Hilbert space that is a **tensor product**:

$$


\mathcal{H}_{\text{total}} = \mathcal{H}_{j_1} \otimes \mathcal{H}_{j_2}

$$


This space has dimension $(2j_1 + 1)(2j_2 + 1)$.

### Coupled and Uncoupled Bases

#### The Two Perspectives

The **uncoupled basis** $|j_1, m_1\rangle |j_2, m_2\rangle$ is an eigenbasis of $\hat{J}_{1z}$, $\hat{J}_{2z}$ (and the squares $\hat{\boldsymbol{J}}_1^2$, $\hat{\boldsymbol{J}}_2^2$). It has $(2j_1 + 1)(2j_2 + 1)$ states.

The **coupled basis** $|J, M\rangle$ is an eigenbasis of $\hat{\boldsymbol{J}}^2_\text{tot}$ and $\hat{J}_{z,\text{tot}}$. It accounts for the fact that the total angular momentum is a single quantum object.

These bases span the same Hilbert space but provide different perspectives:
- Uncoupled: we know the individual angular momenta $j_1, j_2$ and their $z$-components $m_1, m_2$.
- Coupled: we know the magnitude and $z$-component of the total angular momentum $J, M$.

:::{admonition} Triangle Rule
:class: important

When adding angular momenta $j_1$ and $j_2$, the total $J$ ranges from $|j_1 - j_2|$ to $j_1 + j_2$ in integer steps.

$$


|j_1 - j_2| \le J \le j_1 + j_2

$$


For each allowed $J$, the magnetic quantum number ranges as $-J \le M \le J$.

**Physical interpretation**: The total angular momentum must satisfy a triangle inequality—the same rule that governs vectors in classical mechanics.
:::

:::{admonition} Derivation: Triangle Rule
:class: dropdown

The triangle rule $j \in \{|j_1 - j_2|, |j_1 - j_2| + 1, \ldots, j_1 + j_2\}$ follows from the requirement that quantum numbers must be integers or half-integers.

**Key constraint:** For a given value of $j$, the magnetic quantum number must satisfy:

$$


m \in \{-j, -j+1, \ldots, j-1, j\}

$$


In the uncoupled basis, the total $z$-component is:

$$


m_{\text{total}} = m_1 + m_2

$$


**Maximum value of $m_{\text{total}}$:**

$$


m_{\text{max}} = j_1 + j_2

$$


For a coupled state $|j_1, j_2; j, m\rangle$, the maximum value of $m$ is $j$ itself. Therefore:

$$


j_{\text{max}} = j_1 + j_2

$$


**Dimension matching:**

- Uncoupled basis: $(2j_1 + 1)(2j_2 + 1)$ states
- Coupled basis: $\sum_{j} (2j + 1)$ where $j$ ranges over allowed values

These must be equal. With the triangle rule, we get exactly the right count:

$$


\sum_{j=|j_1-j_2|}^{j_1+j_2} (2j + 1) = (2j_1 + 1)(2j_2 + 1) \quad ✓

$$


Thus the minimum is $j_{\text{min}} = |j_1 - j_2|$, giving the full range.
:::

### Clebsch-Gordan Coefficients: Change of Basis

:::{admonition} Clebsch-Gordan Coefficients
:class: important

The **Clebsch-Gordan coefficients** $\langle j_1, m_1; j_2, m_2 | J, M \rangle$ are the overlaps that express the coupled basis in terms of the uncoupled basis:

$$


|J, M\rangle = \sum_{m_1, m_2} \langle j_1, m_1; j_2, m_2 | J, M \rangle |j_1, m_1\rangle |j_2, m_2\rangle

$$


Equivalently:

$$


|j_1, m_1\rangle |j_2, m_2\rangle = \sum_{J, M} \langle J, M | j_1, m_1; j_2, m_2 \rangle |J, M\rangle

$$


This is a **unitary transformation** between the two bases. Clebsch-Gordan coefficients are nonzero only when:
- $M = m_1 + m_2$ (conservation of $m$)
- $|j_1 - j_2| \leq J \leq j_1 + j_2$ (triangle rule)
- The coefficients are **real numbers** (not complex)
:::

#### Properties and Selection Rules

1. **Orthonormality**:

   $$


   \sum_{m_1, m_2} \langle j_1, m_1; j_2, m_2 | J, M \rangle \langle j_1, m_1; j_2, m_2 | J', M' \rangle = \delta_{JJ'} \delta_{MM'}

   $$

2. **Selection rules**: CG coefficients vanish unless $M = m_1 + m_2$ and triangle rule is satisfied.

3. **Computational access**: CG coefficients are typically:
   - **Tabulated** in physics references
   - **Computed** by algebra systems (Mathematica, SymPy)
   - **Evaluated** online via databases (NIST, etc.)

For simple cases, they can be derived algebraically.

:::{admonition} Derivation: Clebsch-Gordan Orthonormality
:class: dropdown

Both $|j_1, m_1; j_2, m_2\rangle$ and $|J, M\rangle$ are orthonormal bases. Expanding using CG coefficients:

$$


|J, M\rangle = \sum_{m_1, m_2} \langle j_1, m_1; j_2, m_2 | J, M \rangle |j_1, m_1; j_2, m_2\rangle

$$


Take the inner product of two such states:

$$


\langle J', M' | J, M \rangle = \sum_{m_1, m_2} \langle j_1, m_1; j_2, m_2 | J', M' \rangle \langle j_1, m_1; j_2, m_2 | J, M \rangle

$$


Since both sides must equal $\delta_{JJ'} \delta_{MM'}$:

$$


\delta_{JJ'} \delta_{MM'} = \sum_{m_1, m_2} \langle j_1, m_1; j_2, m_2 | J', M' \rangle \langle j_1, m_1; j_2, m_2 | J, M \rangle

$$


This is the **orthonormality relation** for CG coefficients.
:::

### Example: Two Spin-1/2 Particles

#### Triangle Rule Applied

For $j_1 = 1/2$ and $j_2 = 1/2$:

$$


J \in \{|1/2 - 1/2|, \ldots, 1/2 + 1/2\} = \{0, 1\}

$$


So we can have:
- **Singlet** ($J=0$): 1 state with $M = 0$
- **Triplet** ($J=1$): 3 states with $M = -1, 0, +1$

Total: $1 + 3 = 4$ states, which matches $(2 \cdot 1/2 + 1)(2 \cdot 1/2 + 1) = 2 \times 2 = 4$. ✓

#### The Four Uncoupled Basis States

$$


|\uparrow\uparrow\rangle = |+\rangle_1 |+\rangle_2, \quad
|\uparrow\downarrow\rangle = |+\rangle_1 |-\rangle_2, \quad
|\downarrow\uparrow\rangle = |-\rangle_1 |+\rangle_2, \quad
|\downarrow\downarrow\rangle = |-\rangle_1 |-\rangle_2

$$

:::{admonition} Triplet States ($J = 1$)
:class: important

The **triplet** ($J = 1$) consists of three **symmetric** combinations:

$$


|1, 1\rangle = |\uparrow\uparrow\rangle

$$


$$

|1, 0\rangle = \frac{1}{\sqrt{2}} (|\uparrow\downarrow\rangle + |\downarrow\uparrow\rangle)

$$


$$

|1, -1\rangle = |\downarrow\downarrow\rangle

$$ (eq-triplet-states)

Each triplet state is **symmetric** under particle exchange: $P_{12} |1, M\rangle = |1, M\rangle$.
:::

:::{admonition} Singlet State ($J = 0$)
:class: important

The **singlet** ($J = 0$) is the **antisymmetric** combination:

$$


|0, 0\rangle = \frac{1}{\sqrt{2}} (|\uparrow\downarrow\rangle - |\downarrow\uparrow\rangle)

$$ (eq-singlet-state)

The singlet state is **antisymmetric** under particle exchange: $P_{12} |0, 0\rangle = -|0, 0\rangle$.
:::

### Physical Implications: Singlet and Triplet States

#### Spin Correlations in the Singlet

The singlet state has a remarkable property:

**If we measure the spin of particle 1 along $z$ and find $\uparrow$, then particle 2 is **guaranteed** to be $\downarrow$** (and vice versa).

This is **maximal entanglement**: the two spins are perfectly anticorrelated.

#### Bell's Inequality and Non-locality

The singlet state can be used to test **Bell's inequality**, demonstrating that quantum mechanics exhibits stronger correlations than any classical hidden-variable theory allows. This is a fundamental test of the non-classical nature of quantum mechanics.

#### In Atomic and Molecular Physics

**Helium atom ground state**: Two electrons in the $1s$ orbital must form a spin singlet (to satisfy the Pauli exclusion principle, as the spatial part is symmetric).

$$


\Psi_{He, \text{ground}} = \phi_{1s}(\boldsymbol{r}_1) \phi_{1s}(\boldsymbol{r}_2) \times |0, 0\rangle

$$


The two electron spins are paired (antiparallel).

**Hydrogen molecule** $H_2$: Two electrons can pair in a singlet state, lowering the energy and stabilizing the bond.

### Spin-Orbit Coupling

#### The Interaction Hamiltonian

An electron moving in an electromagnetic field experiences a **spin-orbit interaction** proportional to:

:::{admonition} Spin-Orbit Coupling
:class: important

$$


H_{SO} = \lambda \boldsymbol{L} \cdot \boldsymbol{S}

$$ (eq-spinorbit-coupling)

where $\boldsymbol{L}$ is orbital angular momentum, $\boldsymbol{S}$ is spin, and $\lambda$ is a coupling constant that depends on the electric field gradient.
:::

#### Why the Coupled Basis is Essential

The spin-orbit interaction $\boldsymbol{L} \cdot \boldsymbol{S}$ is **diagonal** in the coupled basis $|\ell, s; j, m_j\rangle$ but **not** in the uncoupled basis $|\ell, m_\ell; s, m_s\rangle$.

Using coupled basis states simplifies calculations enormously because:
- In the uncoupled basis, the interaction mixes many states.
- In the coupled basis, each state evolves independently.

#### Fine Structure: An Example

For an electron in an atom with orbital angular momentum $\ell$:

$$


j \in \{\ell - 1/2, \ell + 1/2\} \quad \text{(for } \ell \geq 1\text{)}

$$


The energy splitting between $j = \ell + 1/2$ and $j = \ell - 1/2$ is the **fine structure**—a small but measurable correction to atomic energies.

**Example**: Hydrogen $2p$ level splits into $2p_{3/2}$ and $2p_{1/2}$.

### Summary

- **Total angular momentum**: $\boldsymbol{J} = \boldsymbol{J}_1 + \boldsymbol{J}_2$; satisfies the same commutation relations.
- **Uncoupled basis**: $|j_1, m_1; j_2, m_2\rangle$ diagonalizes $J_{1z}, J_{2z}$ but not $J^2$.
- **Coupled basis**: $|j_1, j_2; j, M\rangle$ diagonalizes $J^2, J_z$.
- **Triangle rule**: $j \in \{|j_1 - j_2|, \ldots, j_1 + j_2\}$; determines allowed total angular momenta.
- **Clebsch-Gordan coefficients**: Real unitary transformation between bases; provide the change of basis.
- **Two spin-1/2 particles**: Singlet ($j=0$, antisymmetric) and triplet ($j=1$, symmetric) states with explicit forms.
- **Spin-orbit coupling** $\boldsymbol{L} \cdot \boldsymbol{S}$ is naturally diagonal in coupled basis; causes fine structure in atoms.
