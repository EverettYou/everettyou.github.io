## Lecture Notes

If magnetic monopoles existed—particles carrying only magnetic charge with no electric charge—they would fundamentally alter quantum mechanics itself. Dirac showed that monopole quantization must exist: the product of electric and magnetic charges must be quantized. Yet monopoles introduce gauge singularities (Dirac strings) whose position is arbitrary and unobservable. This section explores monopoles as a test of gauge invariance, topological constraints on charge quantization, and the deep principle that gauge singularities are artifacts of gauge choice.
:::{admonition} Dirac String: A Gauge Singularity
:class: note

Dirac's solution uses a gauge singularity—the "string"—that is not physical. Different gauges place the string differently, but all predict identical physical observables (forces, particle trajectories).
:::


### Definition and Field of a Magnetic Monopole

A **magnetic monopole** is a hypothetical particle carrying magnetic charge $g$ (analogous to electric charge), producing a purely radial magnetic field:

$$

\boldsymbol{B}_{\text{mono}} = \frac{g}{r^2}\hat{r}

$$

This is the electromagnetic dual of the Coulomb electric field. If monopoles existed, Maxwell's equations would be completely symmetric under the duality transformation:

$$

\boldsymbol{E} \to \boldsymbol{B}, \quad \boldsymbol{B} \to -\boldsymbol{E}, \quad e \to g, \quad g \to -e

$$

### Vector Potential with Dirac String

The challenge: a monopole's field cannot be represented globally by $\boldsymbol{B} = \nabla \times \boldsymbol{A}$ using a single-valued vector potential. We must introduce a **Dirac string**—a singular line extending to infinity.

For a monopole at the origin with string along the negative $z$-axis, the vector potential in spherical coordinates is:

$$

\boldsymbol{A} = \frac{g}{r}\frac{1 - \cos\theta}{\sin\theta}\hat{\phi}

$$

Away from the string ($\theta \neq \pi$), this is smooth and gives:

$$

\nabla \times \boldsymbol{A} = \frac{g}{r^2}\hat{r}

$$

the monopole field. At the string ($\theta = \pi$, the negative $z$-axis), the potential diverges—this is where the mathematical singularity sits.

### The Aharonov-Bohm Effect with a Monopole

When a charged particle travels around the monopole (avoiding the string), it acquires a phase:

$$

\phi = \frac{q}{\hbar c}\oint \boldsymbol{A} \cdot d\boldsymbol{l} = \frac{2\pi qg}{\hbar c}

$$

For the wavefunction to be single-valued after encircling the monopole:

$$

e^{i\phi} = 1 \implies \phi = 2\pi n

$$

This gives **Dirac's quantization condition**:

$$

\frac{2\pi qg}{\hbar c} = 2\pi n \quad \Rightarrow \quad qg = n\frac{\hbar c}{2}

$$

This remarkable result connects the Aharonov-Bohm phase (purely quantum) to the topological constraint that wavefunctions must be single-valued on the sphere.


:::{admonition} Derivation
:class: dropdown note


Although the vector potential is singular on the string, a charged particle traveling around the monopole (avoiding the string) acquires a well-defined phase. Consider a path encircling the monopole at the equator ($\theta = \pi/2$):

$$

\phi = \frac{q}{\hbar c}\oint \boldsymbol{A} \cdot d\boldsymbol{l} = \frac{q}{\hbar c} \cdot 2\pi g = \frac{2\pi qg}{\hbar c}

$$

For the wavefunction to be **single-valued** after going around the monopole:

$$

e^{i\phi} = 1 \implies \phi = 2\pi n

$$

This gives the quantization condition:

$$

\frac{2\pi qg}{\hbar c} = 2\pi n

$$

:::
### Monopole Gauge Potential: North and South Patches

Since no single vector potential covers the entire sphere, we use two **coordinate patches** with a gauge transformation between them:

**North patch** (singular at south pole):

$$

\boldsymbol{A}^{(N)} = \frac{g}{r}\frac{1 - \cos\theta}{\sin\theta}\hat{\phi}

$$

**South patch** (singular at north pole):

$$

\boldsymbol{A}^{(S)} = -\frac{g}{r}\frac{1 + \cos\theta}{\sin\theta}\hat{\phi}

$$

In each patch: $\nabla \times \boldsymbol{A} = (g/r^2)\hat{r}$ (away from singularity).

In the overlap region, the patches are related by a large gauge transformation with winding number. The wavefunction must satisfy:

$$

\psi^{(N)}|_{\text{overlap}} = e^{-i\chi}\psi^{(S)}|_{\text{overlap}}, \quad \chi = \frac{2g}{\hbar c}\phi

$$

Single-valuedness of the wavefunction enforces the quantization condition $qg = n(\hbar c/2)$.


:::{admonition} Derivation
:class: dropdown note

Since no single-valued vector potential covers the entire sphere, Dirac's solution uses two **coordinate patches**:
- **North patch**: covers $0 \leq \theta \leq \pi/2 + \epsilon$
- **South patch**: covers $\pi/2 - \epsilon \leq \theta \leq \pi$

**North patch potential** ($\text{monopole on negative z-axis}$):

$$

\boldsymbol{A}^{(N)} = \frac{g}{r}\frac{1 - \cos\theta}{\sin\theta}\hat{\phi}

$$

This is singular at $\theta = \pi$ (the south pole), where the Dirac string terminates.

**South patch potential**:

$$

\boldsymbol{A}^{(S)} = -\frac{g}{r}\frac{1 + \cos\theta}{\sin\theta}\hat{\phi}

$$

This is singular at $\theta = 0$ (the north pole), where the string would re-emerge if we extended it that way.

**Verification of curl**: In each patch, we can verify:

$$

\nabla \times \boldsymbol{A}^{(N)} = \frac{g}{r^2}\hat{r}

$$

(away from the string).

**Gauge transformation relating patches**: In the overlap region, the north and south patch potentials differ by a gauge transformation:

$$

\boldsymbol{A}^{(N)} - \boldsymbol{A}^{(S)} = \nabla\chi

$$

with $\chi = \frac{2g}{\hbar c}\phi$ (up to a normalization constant). This shows that the two patches are related by a large gauge transformation—a U(1) transformation with winding number.

The wavefunction in the overlap region must satisfy:

$$

\psi^{(N)}|_{\text{overlap}} = e^{-i\chi}\psi^{(S)}|_{\text{overlap}}

$$

This boundary condition enforces that the particle's wavefunction is single-valued only if the Dirac quantization condition $qg = n(\hbar c/2)$ is satisfied.

:::

### Dirac's Quantization Condition

**Dirac's Quantization Condition** (1931):

$$

qg = n\frac{\hbar c}{2}, \quad n = 0, \pm 1, \pm 2, \ldots

$$

This is one of the most profound results in theoretical physics. Even though monopoles have never been observed, the existence of even a single monopole anywhere in the universe would imply that **electric charge is quantized**.

More remarkably, if we set $q = e$ (the electron charge) and ask what magnetic charges are allowed, we find:

$$

g = n\frac{\hbar c}{2e}

$$

The magnetic charge is quantized in units of $\hbar c/(2e)$, the **fundamental magnetic charge quantum**.

### Physical Interpretation

Dirac's result suggests that **charge quantization is a consequence of the existence of monopoles** in the universe, even if they're rare or confined to unreachable regions. The discovery that all observed particles have electric charges that are integer multiples of $e$ is thereby explained by topology rather than accident.

In modern particle physics, monopoles appear naturally in certain grand unified theories (GUTs) and in certain condensed matter systems. Although elusive in high-energy physics, monopole-like excitations do exist in materials with topological order.

### Gauge Transformation and String Invisibility

A crucial point: the Dirac string is **gauge-dependent**. Different gauge choices place the string in different locations. While the string location changes, all physical observables (wavefunction phases, forces, etc.) remain invariant.

Under a gauge transformation to a new vector potential $\boldsymbol{A}' = \boldsymbol{A} - \nabla\chi$, the string shifts but the physics stays the same. This demonstrates that the string is not physical—it's an artifact of choosing a specific gauge.

### Monopole in Quantum Mechanics: The Sphere Geometry

When a charged particle is restricted to move on a sphere in the presence of a monopole at the center, the problem becomes a quantum mechanics on a curved manifold with a non-trivial gauge field—the monopole background.

The wavefunction $\psi(\theta, \phi)$ must be single-valued on the sphere, and the boundary condition from the monopole's Aharonov-Bohm phase constrains the allowed quantum numbers.

### Summary

- **Monopole field**: $\boldsymbol{B} = (g/r^2)\hat{r}$—radial, analogous to electric field of charge.
- **Dirac string**: Singular line extending to infinity; carries unobservable flux.
- **Quantization condition**: $qg = n(\hbar c/2)$ connects electric and magnetic charges topologically.
- **Charge quantization origin**: Monopole existence would explain discrete nature of observed electric charges.
- **Aharonov-Bohm phase**: Particle encircling monopole acquires phase $2\pi qg/(\hbar c)$.
- **Single-valuedness**: Wavefunction must return to itself after $2\pi$ motion around monopole.
- **Gauge freedom**: String location is gauge-dependent; physics is gauge-invariant.
- **Grand Unification**: Monopoles appear naturally in GUT theories.
- **Topological quantization**: Charge quantization is topological consequence, not fundamental axiom.
