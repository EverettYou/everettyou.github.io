## Lecture Notes

### Overview

A charged particle confined to a ring with a perpendicular magnetic flux provides a concrete, fully solvable model of the Aharonov-Bohm effect. The energy spectrum becomes periodic in flux, tunneling rates depend on the enclosed flux, and the ground state carries a persistent currentâ€”a surprising quantum phenomenon with no classical analog and direct applications in superconducting devices.

### The Hamiltonian: Particle on a Ring

Consider a particle of mass $m$ and charge $q$ confined to move on a ring of radius $R$ in the $xy$-plane. A uniform magnetic field perpendicular to the ring creates an enclosed magnetic flux $\Phi$ through the ring's interior.

Using angular coordinate $\theta$ (with $0\leq\theta<2\pi$), the canonical angular momentum is

$$
\hat{L}_{\theta} = -\mathrm{i}\hbar\,\frac{\partial}{\partial\theta}.
$$

In a perpendicular magnetic field, the kinetic angular momentum is shifted by the vector-potential contribution (Peierls substitution), and the Hamiltonian becomes:

:::{admonition} The Flux Ring Hamiltonian
:class: important

$$
\hat{H}(\Phi) = \frac{1}{2mR^{2}}\left(\hat{L}_{\theta} - \frac{q\Phi}{2\pi}\right)^{2},
$$

where $\Phi$ is the total magnetic flux threading the ring. The shift $q\Phi/(2\pi)$ is the dimensionless angular-momentum offset induced by the flux.
:::

### Eigenstate and Eigenenergy

:::{margin}
:::{figure} ../assets/figs/4-2-3-flux_ring_energy.png
:width: 85%
:alt: Flux-ring energy branches E_n versus flux Phi/Phi_0, showing periodic parabolic bands and crossings at half-integer flux

Flux-ring energy branches $E_n(\Phi)$ versus $\Phi/\Phi_0$.
:::
:::

The Hamiltonian commutes with $\hat{L}_{\theta}$, so we can label simultaneous eigenstates of $\hat{L}_{\theta}$ and $\hat{H}$ by the **angular-momentum quantum number** $n$,

$$
\hat{L}_{\theta}\,|n\rangle = \hbar n\,|n\rangle.
$$

Single-valuedness on the ring, $\psi(\theta+2\pi)=\psi(\theta)$, quantizes $n\in\mathbb{Z}$ to integers. The corresponding wavefunctions are plane waves,

$$
\psi_n(\theta) = \langle\theta|n\rangle = \frac{1}{\sqrt{2\pi}}\,\mathrm{e}^{\mathrm{i}n\theta}.
$$

In the angular position basis $\{|\theta\rangle:\theta\in[0,2\pi)\}$, the same state reads

$$
\begin{split}
|n\rangle & \;=\; \int_{0}^{2\pi}\psi_n(\theta)|\theta\rangle\,\mathrm{d}\theta\\
& \;=\; \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}\mathrm{e}^{\mathrm{i}n\theta}|\theta\rangle\,\mathrm{d}\theta.
\end{split}
$$ (eq-ring-ndef)

The energies follow by direct substitution. With $\hat{L}_{\theta}|n\rangle = \hbar n\,|n\rangle$, the flux-dependent Hamiltonian satisfies
$\hat{H}(\Phi)\,|n\rangle = E_n(\Phi)|n\rangle$, giving the eigenvalue

:::{admonition} Flux Ring Energy Spectrum
:class: important

$$
E_n(\Phi) = \frac{\hbar^2}{2mR^2}\left(n - \frac{\Phi}{\Phi_0}\right)^2,
$$ (eq-ring-spectrum)

where $\Phi_0$ is the flux quantum (defined in **[4.2.2 Aharonov-Bohm Effect](4-2-2-aharonov-bohm-effect)**). Shifting flux by one quantum relabels levels but preserves the spectrum,

$$
E_n(\Phi + \Phi_0) = E_{n-1}(\Phi),
$$

so the *set* $\{E_n(\Phi)\}_{n\in\mathbb{Z}}$ is periodic in $\Phi$ with period $\Phi_0$.
:::


### Ground State and Persistent Current

:::{margin}
:::{figure} ../assets/figs/4-2-3-flux_ring_current.png
:width: 100%
:alt: Flux-ring ground-state energy and persistent current versus flux Phi/Phi_0, showing sawtooth current and periodicity

Ground-state response: $E_0(\Phi)$ and sawtooth persistent current $I(\Phi)$.
:::
:::

The ground state at flux $\Phi$ is the eigenstate $|n_*\rangle$ whose quantum number is the integer closest to $\Phi/\Phi_0$. The flux axis decomposes into branches: for $\Phi/\Phi_0\in(n-1/2,\,n+1/2)$ the ground state has quantum number $n$, with adjacent branches degenerate at every half-integer flux. As flux is tuned through any half-integer value the ground-state quantum number jumps by one â€” so although the *spectrum* is periodic in flux, the *ground-state momentum* is not.

:::{admonition} Persistent Current
:class: important

The persistent current is the response of the ground state energy to flux,

$$
I(\Phi) \;=\; -\frac{\partial E_0(\Phi)}{\partial\Phi}.
$$ (eq-persistent-current)

On the branch $\Phi/\Phi_0\in(n-1/2,\,n+1/2)$ where the ground state has quantum number $n$,

$$
I(\Phi) \;=\; \frac{\hbar^{2}}{mR^{2}\,\Phi_0}\!\left(n-\frac{\Phi}{\Phi_0}\right).
$$

It flows continuously in the ground state at zero temperature with no applied voltage, opposes the applied flux (Lenz-like), and reverses sign each time the flux crosses a half-integer â€” giving the sawtooth $I(\Phi)$ characteristic of mesoscopic flux rings.
:::

In a superconducting ring (zero resistance), the persistent current is truly permanent and can be measured via the magnetic field it generates.

### Tunneling Across the Ring: Flux Dependence

Now suppose the ring has a potential barrier that allows tunneling from one side to the other. The tunneling rate depends on the phase accumulated along different paths.

For a path that goes halfway around the ring and acquires a flux-dependent phase:

$$
\Phi_{\mathrm{path}} = \frac{q\Phi}{2\hbar} = \frac{\pi\Phi}{\Phi_0}
$$

The tunneling amplitude oscillates with flux, reaching maximum when $\Phi = m\Phi_0$ (integer multiples) and minimum when $\Phi = (m+1/2)\Phi_0$.

:::{admonition} Aharonov-Bohm Effect in the Flux Ring
:class: note

The tunneling rate $\Gamma(\Phi) = \Gamma_{0}\cos^{2}(\pi\Phi/\Phi_{0})$ oscillates with period $\Phi_{0}$ â€” it is the squared modulus of the tunneling amplitude $\mathcal{T}\propto\cos(\pi\Phi/\Phi_{0})$, so $\Gamma\propto\vert\mathcal{T}\vert^{2}$. This direct observation of flux modulation is a hallmark AB effect: the particle's dynamics depend on the enclosed flux even though the particle never enters the field region.
:::

### Half-Integer Flux: Anomalous Degeneracy

At every half-integer flux $\Phi=(n+1/2)\Phi_0$ two adjacent branches of the spectrum cross. The simplest case is $\Phi = \Phi_0/2$, where $|0\rangle$ and $|1\rangle$ become degenerate.

:::{admonition} The half-flux doublet
:class: important

Specializing {eq}`eq-ring-ndef` to $n=0$ and $n=1$,

$$
|0\rangle \;=\; \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}|\theta\rangle\,\mathrm{d}\theta,
\qquad
|1\rangle \;=\; \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}\mathrm{e}^{\mathrm{i}\theta}|\theta\rangle\,\mathrm{d}\theta.
$$

$|0\rangle$ is uniform amplitude on the ring (no winding); $|1\rangle$ carries one full winding of phase. At $\Phi = \Phi_0/2$ both have the same energy:

$$
E_0(\Phi_0/2) \;=\; E_1(\Phi_0/2) \;=\; \frac{\hbar^{2}}{8mR^{2}}.
$$

This degeneracy is not a coincidence â€” it is forced by a **quantum anomaly**.
:::

**The anomaly.** Two symmetries protect the doublet at $\Phi = \Phi_0/2$:

- **Translation by $\pi$**, $\hat{T}_\pi:\theta\to\theta+\pi$, multiplies $|n\rangle$ by $(-1)^n$, so $\hat{T}_\pi|0\rangle = +|0\rangle$ and $\hat{T}_\pi|1\rangle = -|1\rangle$ â€” Pauli $\hat{Z}$.
- **Reflection with large gauge**, $\hat{S} = \hat{R}\,\hat{U}_{LG}$, with $\hat{R}:\theta\to-\theta$ and $\hat{U}_{LG}\psi(\theta) = \mathrm{e}^{-\mathrm{i}\theta}\psi(\theta)$ (shifts the flux by one quantum); on the doublet $\hat{S}|0\rangle = |1\rangle$ and $\hat{S}|1\rangle = |0\rangle$ â€” Pauli $\hat{X}$.

As geometric operations on the ring both are abelian, but their lift to the Hilbert space picks up an extra phase $\mathrm{e}^{\mathrm{i}\pi}=-1$ that makes them **anticommute**: $\{\hat{T}_\pi,\hat{S}\}=0$. This extra phase *is* the anomaly. The smallest unitary representation of two anticommuting operators is two-dimensional, so any perturbation preserving both symmetries cannot lift the doublet.

:::{admonition} What is a quantum anomaly?
:class: dropdown note

A **quantum anomaly** is a symmetry that closes naively at the classical level but acquires an extra phase when lifted to operators on the Hilbert space. Here, $\hat{T}_\pi$ and the geometric part of $\hat{S}$ are abelian as ring transformations: translating then reflecting and reflecting then translating give the same point. But once they act on quantum states the lift carries an extra phase $\mathrm{e}^{\mathrm{i}\pi}=-1$, turning ordinary commutation into anticommutation.

This *projective* realization of the symmetry algebra is called a **'t Hooft anomaly**. Its consequence here is sharp: the smallest faithful representation is forced to be larger than 1-dimensional, so a non-degenerate ground state is *forbidden* as long as both symmetries are intact.

Anomalies of this kind are central to modern condensed matter and quantum field theory: they classify symmetry-protected topological phases, protect the gapless edge modes of topological insulators, and constrain the low-energy spectra of strongly-correlated systems. The flux ring is among the simplest concrete examples â€” accessible without quantum field theory yet capturing the same essential mechanism.
:::

:::{admonition} Derivation: Symmetry Action on the Doublet
:class: dropdown information

Each symmetry $\hat{O}$ satisfies $\hat{O}\hat{H}(\Phi_0/2)\hat{O}^{-1}=\hat{H}(\Phi_0/2)$. We compute its action using $|n\rangle = (1/\sqrt{2\pi})\int_{0}^{2\pi}\mathrm{e}^{\mathrm{i}n\theta}|\theta\rangle\,\mathrm{d}\theta$.

**Translation by $\pi$.** Define $\hat{T}_\pi|\theta\rangle = |\theta+\pi\rangle$. Then

$$
\begin{split}
\hat{T}_\pi|n\rangle
&= \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}\mathrm{e}^{\mathrm{i}n\theta}|\theta+\pi\rangle\,\mathrm{d}\theta\\
&= \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}\mathrm{e}^{\mathrm{i}n(\theta'-\pi)}|\theta'\rangle\,\mathrm{d}\theta'\\
&= \mathrm{e}^{-\mathrm{i}n\pi}|n\rangle = (-1)^{n}|n\rangle,
\end{split}
$$

(substituting $\theta'=\theta+\pi$ and using $2\pi$-periodicity to keep the integration limits). On the doublet, $\hat{T}_\pi=\mathrm{diag}(+1,-1) = \hat{Z}$.

**Reflection.** Define $\hat{R}|\theta\rangle = |-\theta\rangle$. Then

$$
\hat{R}|n\rangle = \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}\mathrm{e}^{\mathrm{i}n\theta}|-\theta\rangle\,\mathrm{d}\theta = \frac{1}{\sqrt{2\pi}}\int_{0}^{2\pi}\mathrm{e}^{-\mathrm{i}n\theta'}|\theta'\rangle\,\mathrm{d}\theta' = |-n\rangle.
$$

So $\hat{R}|0\rangle=|0\rangle$ but $\hat{R}|1\rangle=|-1\rangle$ â€” outside the doublet. Reflection alone is not a symmetry of $\hat{H}(\Phi_0/2)$ either: $\hat{R}\hat{H}(\Phi)\hat{R}^{-1}=\hat{H}(-\Phi)$, since $\hat{R}\hat{L}_\theta\hat{R}^{-1}=-\hat{L}_\theta$ flips the canonical momentum (equivalent to flipping the flux).

**Large gauge transformation.** $\hat{U}_{LG}$ multiplies wavefunctions by $\mathrm{e}^{-\mathrm{i}\theta}$:

$$
\hat{U}_{LG}\psi_{n}(\theta) = \mathrm{e}^{-\mathrm{i}\theta}\,\frac{\mathrm{e}^{\mathrm{i}n\theta}}{\sqrt{2\pi}} = \psi_{n-1}(\theta), \qquad \hat{U}_{LG}|n\rangle = |n-1\rangle.
$$

This shifts the canonical-momentum eigenvalue $\hbar n\to\hbar(n-1)$. Since $\hat{L}_\theta$ and $\Phi$ enter the Hamiltonian only through the combination $\hat{L}_\theta - q\Phi/(2\pi)$, the same shift is equivalent to advancing the flux, $\Phi\to(\Phi-\Phi_0)$, (using $q\Phi_0/(2\pi)=\hbar$). Hence $\hat{U}_{LG}\hat{H}(\Phi)\hat{U}_{LG}^{-1}=\hat{H}(\Phi-\Phi_0)$ â€” this is why $\hat{U}_{LG}$ is called a *large* gauge transformation: it changes the flux by a full quantum, an operation that small (continuous) gauge transformations cannot accomplish.

**Combination $\hat{S}=\hat{R}\,\hat{U}_{LG}$.** Acting on $|n\rangle$,

$$
\hat{S}|n\rangle = \hat{R}\hat{U}_{LG}|n\rangle = \hat{R}|n-1\rangle = |-(n-1)\rangle = |1-n\rangle.
$$

On the doublet, $\hat{S}|0\rangle=|1\rangle$ and $\hat{S}|1\rangle=|0\rangle$ â€” Pauli $\hat{X}$. As a symmetry check,

$$
\hat{S}\hat{H}(\Phi)\hat{S}^{-1} = \hat{R}\hat{H}(\Phi-\Phi_0)\hat{R}^{-1} = \hat{H}(\Phi_0-\Phi),
$$

which equals $\hat{H}(\Phi_0/2)$ when $\Phi=\Phi_0/2$.

**Anticommutation.** On a general $|n\rangle$,

$$
\hat{T}_\pi\hat{S}|n\rangle = \hat{T}_\pi|1-n\rangle = (-1)^{1-n}|1-n\rangle,\qquad \hat{S}\hat{T}_\pi|n\rangle = \hat{S}\,(-1)^{n}|n\rangle = (-1)^{n}|1-n\rangle.
$$

The ratio $(-1)^{1-n}/(-1)^{n} = (-1)^{1-2n} = -1$, so $\hat{T}_\pi\hat{S} = -\hat{S}\hat{T}_\pi$ on every $|n\rangle$, hence on the entire Hilbert space.
:::

### Application â€” The SQUID

A **SQUID** (Superconducting Quantum Interference Device) replaces the single-particle ring of this section with a superconducting loop containing two Josephson junctions. The loop carries a macroscopic Cooper-pair wavefunction with a well-defined phase, and the two arms of the loop act as the two paths of an Aharonov-Bohm interferometer.

:::{figure} ../assets/figs/4-2-3-squid.png
:width: 50%
:alt: DC SQUID schematic: superconducting ring with two Josephson junctions, bias current leads, and magnetic flux threading the loop

Schematic dc SQUID: bias current splits through two arms that meet at Josephson junctions, while magnetic flux $\Phi$ threads the loop.
:::

Two features of the flux ring carry over directly:

- **AB phase from enclosed flux.** Cooper pairs traversing the two arms pick up a relative phase $2\pi\Phi/\Phi_0$ â€” the same flux-controlled phase that modulates the tunneling rate of the single-particle ring, but now between two macroscopically occupied superconducting paths.
- **Superconducting flux quantum.** The carriers are Cooper pairs of charge $q = 2e$, so $\Phi_0 = h/(2e)$ â€” half the electron flux quantum. SQUID interference repeats with this *superconducting* flux quantum.

Sweeping the external flux modulates the maximum supercurrent the device can sustain: constructive interference at integer $\Phi/\Phi_0$, destructive interference at half-integer. Because tiny flux changes cause measurable changes in the supercurrent, the SQUID is among the most sensitive magnetometers in existence â€” capable of detecting changes as small as $10^{-6}\Phi_0$. The detailed derivation of the modulation involves the Josephson current-phase relation $I = I_c\sin\delta$ at each junction together with flux quantization around the loop; we use the $|\cos|$ form of $I_{\max}(\Phi)$ as a known result in the homework 6.

### Summary

- The **flux ring model** is a fully solvable realization of the Aharonov-Bohm effect where a charged particle on a ring is threaded by perpendicular magnetic flux $\Phi$.

- The **energy spectrum** is periodic in the flux quantum $\Phi_0 = h/q$; at **pi-flux** ($\Phi = \Phi_0/2$) the ground state becomes doubly degenerate due to time-reversal and translational symmetry.

- **Persistent currents** ($\propto \partial E_0/\partial\Phi$) and **modulated tunneling rates** manifest the quantum response to enclosed fluxâ€”a phenomenon with no classical analog.

- **SQUIDs** realize the flux ring as a two-path supercurrent interferometer: $I_{\text{max}}(\Phi) = 2I_c|\cos(\pi\Phi/\Phi_0)|$, making them ultrasensitive magnetometers.
