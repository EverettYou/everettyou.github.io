## Lecture Notes

In quantum mechanics, a charged particle in a uniform magnetic field cannot have arbitrarily small or large energy—its kinetic energy is quantized into discrete Landau levels separated by ℏωc. This quantization is a purely quantum effect with no classical analog, yet it arises directly from solving the Schrödinger equation. Landau quantization is the gateway to understanding the quantum Hall effect, topological properties of energy bands, and how symmetry constrains the quantum mechanical spectrum.
### Quantum Hamiltonian in Symmetric Gauge

When a quantum particle moves in a uniform magnetic field, the Hamiltonian with minimal coupling to the vector potential is:

$$

\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - e\boldsymbol{A})^2

$$

In the **symmetric gauge** (which preserves rotational symmetry), the vector potential is $\boldsymbol{A} = \frac{1}{2}\boldsymbol{B} \times \boldsymbol{r}$. The kinetic momentum components don't commute:

$$

[\pi_x, \pi_y] = -i\hbar\omega_c

$$

where $\omega_c = eB/m$ is the cyclotron frequency. This non-commuting structure is the key to quantization—it signals that the system is fundamentally quantum and cannot be treated classically.

:::{admonition} Derivation: Kinetic Momentum and Commutation Relations
:class: dropdown note


For $\boldsymbol{B} = B\hat{z}$ and symmetric gauge $\boldsymbol{A} = \frac{B}{2}(-y, x, 0)$, the kinetic momentum is:

$$

\pi_x = p_x + \frac{eBy}{2}, \quad \pi_y = p_y - \frac{eBx}{2}

$$

Computing the commutator using $[p_i, x_j] = -i\hbar\delta_{ij}$:

$$

[\pi_x, \pi_y] = [p_x, -\frac{eBx}{2}] + [\frac{eBy}{2}, p_y] = -i\hbar\omega_c

$$

:::

### Harmonic Oscillator Ladder Operators

The non-commuting kinetic momentum components signal a **harmonic oscillator structure** underlying the problem. We can construct ladder operators analogous to the quantum harmonic oscillator.

First, define the **magnetic length**—a fundamental length scale:

$$

\ell_B = \sqrt{\frac{\hbar}{eB}}

$$

Then define the **ladder operators**:

$$

\hat{a} = \frac{1}{\sqrt{2}\hbar\omega_c}(\pi_x + i\pi_y), \quad \hat{a}^\dagger = \frac{1}{\sqrt{2}\hbar\omega_c}(\pi_x - i\pi_y)

$$

These satisfy the **canonical commutation relation**:

$$

[\hat{a}, \hat{a}^\dagger] = 1

$$

This is the hallmark of a harmonic oscillator. The ladder operators allow us to construct energy eigenstates by starting from the ground state and applying creation operators.

:::{admonition} Derivation: Proving the Canonical Commutation
:class: dropdown note


$$

[\hat{a}, \hat{a}^\dagger] = \frac{1}{2\hbar\omega_c^2}[(\pi_x + i\pi_y), (\pi_x - i\pi_y)]

$$

$$

= \frac{1}{2\hbar\omega_c^2}[-2i[\pi_x, \pi_y]] = \frac{1}{2\hbar\omega_c^2}[-2i(-i\hbar\omega_c)] = 1

$$

:::

### Landau Level Hamiltonian

The kinetic energy can be rewritten in terms of the ladder operators:

$$

\hat{H} = \hbar\omega_c\left(\hat{a}^\dagger\hat{a} + \frac{1}{2}\right) = \hbar\omega_c\left(\hat{N} + \frac{1}{2}\right)

$$

where $\hat{N} = \hat{a}^\dagger\hat{a}$ is the occupation number operator. The energy eigenvalues are:

$$

E_n = \hbar\omega_c\left(n + \frac{1}{2}\right), \quad n = 0, 1, 2, \ldots

$$

These are the **Landau level energies**. Each level is labeled by $n$ and has **equal spacing** $\hbar\omega_c$. Even the ground state ($n=0$) has finite zero-point energy $E_0 = \hbar\omega_c/2$—particles cannot be at rest in a magnetic field; they must undergo quantum zero-point cyclotron motion. This is radically different from classical mechanics.

:::{admonition} Derivation: Hamiltonian in Ladder Operator Form
:class: dropdown note


The kinetic energy $\frac{1}{2m}(\pi_x^2 + \pi_y^2)$ can be expressed as:

$$

\frac{1}{2m}(\pi_x^2 + \pi_y^2) = \frac{\hbar\omega_c}{2}(\hat{a}^\dagger\hat{a} + \hat{a}\hat{a}^\dagger)

$$

Using $\hat{a}\hat{a}^\dagger = \hat{a}^\dagger\hat{a} + 1$:

$$

\frac{1}{2m}(\pi_x^2 + \pi_y^2) = \hbar\omega_c\left(\hat{a}^\dagger\hat{a} + \frac{1}{2}\right)

$$

The total Hamiltonian follows directly.

:::

### Landau Level Wavefunctions

The quantum states are characterized by two quantum numbers: the **Landau level index** $n$ and the **guiding center position** $(X_c, Y_c)$—the center of the particle's cyclotron orbit. For the lowest Landau level ($n=0$), the ground-state wavefunction in symmetric gauge is:

$$

\psi_0(x, y) = \frac{1}{(\pi\ell_B^2)^{1/4}}\exp\left(-\frac{x^2+y^2}{2\ell_B^2}\right)

$$

This is a **Gaussian** with width set by the magnetic length $\ell_B$. States with different guiding centers are translations of this wavefunction. Higher Landau levels ($n > 0$) contain additional structure—the wavefunction oscillates within the Gaussian envelope via Hermite polynomials:

$$

\psi_n(x,y) \propto H_n\left(\frac{\sqrt{x^2+y^2}}{\ell_B}\right)\exp\left(-\frac{x^2+y^2}{2\ell_B^2}\right)

$$

The **magnetic length** $\ell_B$ is the characteristic scale: wavefunctions are localized within a region of size $\ell_B$, independent of the particle's energy. This is a purely quantum effect with no classical analog.

:::{admonition} Derivation: Wavefunction Structure and Hermite Polynomials
:class: dropdown note


The Landau level states can be labeled by $|n; X_c, Y_c\rangle$, where $n$ is the Landau level and $(X_c, Y_c)$ is the guiding center. In symmetric gauge, the ground state is the state annihilated by $\hat{a}$:

$$

\hat{a}|0; X_c, Y_c\rangle = 0

$$

Solving this equation in position space gives the Gaussian form. Excited Landau level states are obtained by applying the creation operator $\hat{a}^\dagger$:

$$

|n; X_c, Y_c\rangle = \frac{(\hat{a}^\dagger)^n}{\sqrt{n!}}|0; X_c, Y_c\rangle

$$

The action of $\hat{a}^\dagger$ introduces Hermite polynomials in the radial variable.

:::

### Explicit Landau Level Wavefunctions: n=0,1,2

While the general form uses Hermite polynomials, the first few Landau levels have particularly simple and instructive forms. Let's write them out explicitly for the lowest three Landau levels.

**Lowest Landau Level ($n=0$):**

$$

\psi_0(x, y) = \frac{1}{(\pi\ell_B^2)^{1/4}}\exp\left(-\frac{x^2+y^2}{2\ell_B^2}\right)

$$

This is a **pure Gaussian**—the simplest form. All states within the lowest Landau level are obtained by translating this wavefunction to different guiding center positions $(X_c, Y_c)$.

**First Excited Landau Level ($n=1$):**

$$

\psi_1(x, y) = \frac{1}{(\pi\ell_B^2)^{1/4}} \cdot \sqrt{2} \cdot \frac{\sqrt{x^2+y^2}}{\ell_B} \exp\left(-\frac{x^2+y^2}{2\ell_B^2}\right)

$$

This includes a factor of $\sqrt{x^2+y^2}$ (the radial coordinate multiplied by the Hermite polynomial $H_1$), giving it a dipole-like shape. The wavefunction vanishes at the origin.

**Second Excited Landau Level ($n=2$):**

$$

\psi_2(x, y) = \frac{1}{(\pi\ell_B^2)^{1/4}} \cdot \frac{1}{\sqrt{2}} \left[2\left(\frac{x^2+y^2}{\ell_B^2}\right) - 1\right] \exp\left(-\frac{x^2+y^2}{2\ell_B^2}\right)

$$

This includes a radial polynomial (related to $H_2$), creating a quadrupole-like structure with a radial node.

### Complex Coordinate Formalism

A powerful technique for analyzing Landau level wavefunctions uses **complex coordinates** on the plane. Define:

$$

z = x + iy, \quad \bar{z} = x - iy

$$

Then $x = (z + \bar{z})/2$ and $y = (z - \bar{z})/(2i)$. In these complex variables, the **lowering operator** $\hat{a}$ becomes:

$$

\hat{a} \propto \frac{\partial}{\partial \bar{z}} + \frac{z}{2\ell_B^2}

$$

This reveals a deep fact: in the symmetric gauge, the Landau level wavefunctions are **holomorphic functions** times a Gaussian. Specifically, the $n$-th Landau level wavefunction can be written as:

$$

\psi_n(z, \bar{z}) = P_n(z)\exp\left(-\frac{|z|^2}{2\ell_B^2}\right)

$$

where $P_n(z)$ is a **holomorphic polynomial** of degree $n$. For example:
- $n=0$: $P_0(z) = 1$ (constant)
- $n=1$: $P_1(z) = z$ (linear)
- $n=2$: $P_2(z) = z^2$ (quadratic)

This holomorphic structure is not accidental—it reflects the deep connection between Landau quantization and **quantum Hall geometry**. The fact that $n$-th Landau level states are spanned by degree-$n$ polynomials, with one state per flux quantum, is the foundation of the modern geometric understanding of the quantum Hall effect.

:::{admonition} Why Holomorphic? The Connection to Ladder Operators
:class: dropdown note

The lowering operator $\hat{a}$ in the symmetric gauge can be written as:

$$

\hat{a} \propto \ell_B^{-1}\left(\frac{\partial}{\partial \bar{z}} + \frac{z}{2\ell_B^2}\right)

$$

The ground state satisfies $\hat{a}|0\rangle = 0$, which in position space is:

$$

\left(\frac{\partial}{\partial \bar{z}} + \frac{z}{2\ell_B^2}\right)\psi_0(z, \bar{z}) = 0

$$

The solution is a function with **no $\bar{z}$-dependence** (after accounting for the Gaussian), i.e., it depends only on $z$:

$$

\psi_0(z, \bar{z}) = \exp\left(-\frac{|z|^2}{2\ell_B^2}\right)

$$

Excited Landau levels are obtained by $\psi_n = (\hat{a}^\dagger)^n\psi_0$. Since $\hat{a}^\dagger \propto \partial/\partial z$, each application of the creation operator multiplies the wavefunction by $z$, giving:

$$

\psi_n(z, \bar{z}) \propto z^n \exp\left(-\frac{|z|^2}{2\ell_B^2}\right)

$$

This is why the $n$-th Landau level is spanned by **holomorphic polynomials** of degree $n$ times the Gaussian.

:::

### Magnetic Length: The Fundamental Scale

The **magnetic length** $\ell_B = \sqrt{\hbar/(eB)}$ sets the only length scale in the Landau problem. It characterizes:

1. **Wavefunction size**: All Landau level wavefunctions are localized within a region of linear size $\sim\ell_B$.

2. **Guiding center spacing**: States are spaced one guiding center per flux quantum, which occupies area $\sim 2\pi\ell_B^2$.

3. **Kinetic energy scale**: The spacing between Landau levels is $\hbar\omega_c = \hbar(eB/m) = (\hbar^2)/(m\ell_B^2)$.

4. **Quantum geometry**: The noncommuting guiding center coordinates $[\hat{X}, \hat{Y}] = -i\ell_B^2$ imply that guiding center space itself is **quantized**, with minimum area cell $\ell_B^2$.

Stronger magnetic fields make $\ell_B$ smaller, confining particles to tighter orbits and pushing the Landau level energies higher. Weaker fields make $\ell_B$ larger, allowing particles to spread out more. The magnetic length thus controls the transition between classical (large $\ell_B$) and quantum (small $\ell_B$) behavior.


### Guiding Center Coordinates

A remarkable feature emerges when we study the center of the cyclotron orbit. Define the **guiding center coordinates**:

$$

\hat{X} = \hat{x} + \frac{\hat{p}_y}{eB}, \quad \hat{Y} = \hat{y} - \frac{\hat{p}_x}{eB}

$$

These represent the actual center of the orbit. Unlike ordinary position coordinates, the guiding center coordinates have a **non-zero commutator**:

$$

[\hat{X}, \hat{Y}] = -i\ell_B^2

$$

This non-commuting structure is a signature that guiding center coordinates live on a **quantized phase space** with minimum uncertainty $\ell_B$. The area of a unit cell in guiding center space is quantized: roughly one guiding center per magnetic flux quantum. This connects the Landau quantization problem to deep topological geometry (monopole phase space) and explains why the quantum Hall effect is a topological phenomenon.

:::{admonition} Derivation: Guiding Center Commutation and Magnetic Length
:class: dropdown note


In symmetric gauge, the kinetic momentum is:

$$

\hat{\pi}_x = \hat{p}_x + \frac{eB}{2}\hat{y}, \quad \hat{\pi}_y = \hat{p}_y - \frac{eB}{2}\hat{x}

$$

The guiding center coordinates $(\hat{X}, \hat{Y})$ are defined such that they commute with all ladder operators and encode the orbital center. Their commutator is:

$$

[\hat{X}, \hat{Y}] = [\hat{x}, \hat{y}] + \frac{1}{(eB)^2}[\hat{p}_y, \hat{p}_x] = 0 + \frac{i\hbar}{(eB)^2}

$$

$$

= -i\frac{\hbar}{eB} = -i\ell_B^2

$$

where we used $[\hat{p}_y, \hat{p}_x] = i\hbar$. This means each guiding center occupies a quantum area of order $\ell_B^2$.

:::

### Summary

- **Symmetric gauge**: $\boldsymbol{A} = \frac{B}{2}(-y, x, 0)$ maintains rotational symmetry.
- **Kinetic momentum**: Components $\pi_i = p_i + eA_i$ commute via $[\pi_x, \pi_y] = -i\hbar\omega_c$.
- **Ladder operators**: $\hat{a} = (\pi_x + i\pi_y)/(\sqrt{2}\hbar\omega_c)$ satisfy $[\hat{a}, \hat{a}^\dagger] = 1$.
- **Harmonic oscillator structure**: Hamiltonian $\hat{H} = \hbar\omega_c(\hat{a}^\dagger\hat{a} + 1/2)$ is an isotropic 2D oscillator.
- **Landau energies**: $E_n = \hbar\omega_c(n + 1/2)$—equally spaced with gap $\hbar\omega_c$.
- **Magnetic length**: $\ell_B = \sqrt{\hbar/(eB)}$ is the quantum length scale; smaller in stronger fields.
- **Degeneracy**: Each Landau level contains $N_\phi \approx A/(2\pi\ell_B^2)$ states—enormous for macroscopic systems.
- **Localization**: All wavefunctions in a given Landau level are localized on scales $\sim\ell_B$.
- **Guiding center**: Position-dependent, reflecting the fact that different states within a Landau level differ only in guiding center location.
