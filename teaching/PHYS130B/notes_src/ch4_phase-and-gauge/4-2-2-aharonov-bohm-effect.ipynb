## Lecture Notes

### Overview

The Aharonov-Bohm (AB) effect demonstrates that a charged particle's quantum phase depends on the electromagnetic vector potential $\boldsymbol{A}$, even in regions where the magnetic field vanishes. This reveals that the gauge potential is physically real in quantum mechanics—not merely a mathematical convenience. The effect leads directly to **flux quantization**: in systems with a macroscopic wavefunction (superconductors), magnetic flux is restricted to discrete quantum units $\Phi = n\Phi_0$.

### The Aharonov-Bohm Setup

A solenoid produces a uniform magnetic field $\boldsymbol{B}$ inside but zero field outside. An electron beam splits around the solenoid—one path above, one below—then recombines at a screen.

**Classical prediction:** No effect. The electron experiences no Lorentz force ($\boldsymbol{F} = -e\boldsymbol{v} \times \boldsymbol{B} = 0$) in the field-free region, so both paths are identical.

**Quantum reality:** The interference pattern *shifts*. The two beams acquire different quantum phases, even though neither experiences a magnetic force.

For a solenoid of radius $R$ with internal field $B$ along $\hat{\boldsymbol{z}}$, the vector potential outside ($r > R$) in cylindrical coordinates is

$$

\boldsymbol{A}_\text{out} = \frac{\Phi_B}{2\pi r}\hat{\boldsymbol{\theta}}

$$

where $\Phi_B = \pi R^2 B$ is the enclosed flux. The curl vanishes ($\nabla \times \boldsymbol{A} = 0$), yet $\boldsymbol{A} \neq 0$. This is possible because the region is not simply connected—$\boldsymbol{A}$ cannot be written as a gradient globally.

:::{admonition} Aharonov-Bohm Phase Shift
:class: important

A charged particle traveling around a solenoid acquires a phase shift:

$$

\Delta\phi = \frac{q}{\hbar}\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q\Phi_B}{\hbar}

$$

where $\Phi_B$ is the total magnetic flux enclosed by the path. This phase shift occurs despite $\boldsymbol{B} = 0$ along the particle's trajectory.
:::

:::{admonition} Derivation: AB Phase from Minimal Coupling
:class: dropdown information

The phase arises from the gauge-covariant momentum $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$. In a region where $\boldsymbol{B} = 0$, we can locally write $\boldsymbol{A} = \nabla\Lambda$ for some function $\Lambda$. The Schrödinger equation

$$

\frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2\psi = E\psi

$$

is solved by the substitution $\psi(\boldsymbol{r}) = \mathrm{e}^{\mathrm{i}q\Lambda(\boldsymbol{r})/\hbar}\psi_0(\boldsymbol{r})$, where $\psi_0$ satisfies the free Schrödinger equation $\hat{\boldsymbol{p}}^2\psi_0/(2m) = E\psi_0$. To verify, note that

$$

(\hat{\boldsymbol{p}} - q\boldsymbol{A})\,\mathrm{e}^{\mathrm{i}q\Lambda/\hbar}\psi_0 = \mathrm{e}^{\mathrm{i}q\Lambda/\hbar}\bigl(-\mathrm{i}\hbar\nabla + q\nabla\Lambda - q\nabla\Lambda\bigr)\psi_0 = \mathrm{e}^{\mathrm{i}q\Lambda/\hbar}\hat{\boldsymbol{p}}\psi_0

$$

so the gauge potential is completely absorbed into a position-dependent phase. A particle traveling along path $\mathcal{C}$ from $\boldsymbol{r}_1$ to $\boldsymbol{r}_2$ acquires the phase

$$

\phi[\mathcal{C}] = \frac{q}{\hbar}\int_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q}{\hbar}\bigl[\Lambda(\boldsymbol{r}_2) - \Lambda(\boldsymbol{r}_1)\bigr]

$$

For two paths sharing the same endpoints but encircling the solenoid, the difference is

$$

\Delta\phi = \frac{q}{\hbar}\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q}{\hbar}\int_{\mathcal{S}} (\nabla \times \boldsymbol{A}) \cdot \mathrm{d}\boldsymbol{S} = \frac{q\Phi_B}{\hbar}

$$

where in the last step we used Stokes' theorem (the surface $\mathcal{S}$ bounded by the loop must cross the solenoid, where $\boldsymbol{B} \neq 0$). The key subtlety: $\boldsymbol{A} = \nabla\Lambda$ holds *locally* outside the solenoid, but $\Lambda$ is not single-valued around the solenoid, so $\Lambda(\boldsymbol{r}_2) - \Lambda(\boldsymbol{r}_1)$ depends on the path—this is the topological origin of the AB phase.
:::

### Two-Path Interference

In quantum mechanics, a charged particle accumulates phase $\phi[\mathcal{C}] = (q/\hbar)\int_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$ from minimal coupling ($\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$). Two paths around the solenoid acquire different phases:

$$

\Delta\phi = \phi_\text{upper} - \phi_\text{lower} = \frac{q}{\hbar}\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q\Phi_B}{\hbar}

$$

where the combined paths form a closed loop encircling the solenoid. The interference pattern shifts by $\Delta\phi/(2\pi)$ of a fringe spacing.

:::{admonition} Topological Invariance
:class: important

The circulation $\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \Phi_B$ depends only on whether the loop encircles the solenoid (winding number), not on its shape or size. By Stokes' theorem, the circulation equals the enclosed flux $\int \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S}$.
:::

:::{admonition} Discussion
:class: dropdown tip

In classical mechanics, a particle's trajectory depends only on $\boldsymbol{E}$ and $\boldsymbol{B}$ (the Lorentz force). The vector potential $\boldsymbol{A}$ does not appear. In quantum mechanics, $\boldsymbol{A}$ enters the Schrödinger equation via minimal coupling, making it observable in gauge-invariant combinations. Is the AB effect a sign that quantum mechanics is "deeper" than classical mechanics, or simply that we must describe particles by wavefunctions rather than trajectories?
:::

### Gauge Invariance and Physical Meaning

Under a gauge transformation $\boldsymbol{A} \to \boldsymbol{A} + \nabla\chi$, $\psi \to \mathrm{e}^{\mathrm{i}q\chi/\hbar}\psi$, the phase on a single path changes by $(q/\hbar)[\chi(\text{end}) - \chi(\text{start})]$. But the *relative phase* between two interfering paths sharing the same endpoints is gauge-invariant:

$$

\Delta\phi' = \phi'_1 - \phi'_2 = \Delta\phi

$$

:::{admonition} The Observable is Relative Phase
:class: important

The **relative phase difference** between two paths is gauge-invariant and physically measurable. The absolute phase on a single path is gauge-dependent and unobservable. The circulation $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \Phi_B$ is gauge-invariant because it equals the enclosed flux $\int \boldsymbol{B} \cdot \mathrm{d}\boldsymbol{S}$.
:::

:::{admonition} Derivation: Gauge Invariance of the Relative Phase
:class: dropdown information

Under a gauge transformation $\boldsymbol{A} \to \boldsymbol{A}' = \boldsymbol{A} + \nabla\chi$, the wavefunction transforms as $\psi \to \psi' = \mathrm{e}^{\mathrm{i}q\chi/\hbar}\psi$. The phase accumulated along path $\mathcal{C}_k$ from point $P$ to point $Q$ transforms as:

$$

\phi'_k = \frac{q}{\hbar}\int_{\mathcal{C}_k} \boldsymbol{A}' \cdot \mathrm{d}\boldsymbol{l} = \frac{q}{\hbar}\int_{\mathcal{C}_k} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} + \frac{q}{\hbar}\int_{\mathcal{C}_k} \nabla\chi \cdot \mathrm{d}\boldsymbol{l}

$$

The gradient integral is path-independent (since $\chi$ is single-valued):

$$

\int_{\mathcal{C}_k} \nabla\chi \cdot \mathrm{d}\boldsymbol{l} = \chi(Q) - \chi(P)

$$

This is the *same* for both paths $\mathcal{C}_1$ and $\mathcal{C}_2$, so the relative phase is unchanged:

$$

\Delta\phi' = \phi'_1 - \phi'_2 = (\phi_1 + \tfrac{q}{\hbar}[\chi(Q)-\chi(P)]) - (\phi_2 + \tfrac{q}{\hbar}[\chi(Q)-\chi(P)]) = \phi_1 - \phi_2 = \Delta\phi

$$

Equivalently, $\Delta\phi = (q/\hbar)\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$ is manifestly gauge-invariant because $\oint \nabla\chi \cdot \mathrm{d}\boldsymbol{l} = 0$ for any single-valued $\chi$.
:::

:::{admonition} The Vector Potential is Physical in QM
:class: attention

Classical electromagnetism treats $\boldsymbol{A}$ as a mathematical convenience—only $\boldsymbol{E}$ and $\boldsymbol{B}$ are "real." The AB effect demonstrates that in quantum mechanics, gauge-invariant combinations of $\boldsymbol{A}$ (like $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$) have direct observable consequences. This does not contradict gauge invariance; it shows that $\boldsymbol{A}$ encodes physical information that $\boldsymbol{B}$ alone does not capture in multiply connected regions.
:::

### Flux Quantization

A quantum wavefunction must be single-valued: traveling around a closed loop must return $\psi$ to itself (up to a factor $\mathrm{e}^{2\pi\mathrm{i}n}$). For a superconductor, the macroscopic order parameter $\Psi(\boldsymbol{r})$ describing the Cooper pair condensate (charge $q = 2e$) must satisfy:

$$

\oint \nabla\phi \cdot \mathrm{d}\boldsymbol{l} = 2\pi n \quad \Longrightarrow \quad \frac{2e\Phi}{\hbar} = 2\pi n \quad \Longrightarrow \quad \Phi = n\Phi_0

$$

:::{admonition} Flux Quantization
:class: important

Magnetic flux through a superconducting loop is quantized:

$$

\Phi = n\Phi_0, \quad \Phi_0 = \frac{h}{2e} \approx 2.07 \times 10^{-15}\;\text{Wb}

$$

The factor of 2 in $\Phi_0 = h/(2e)$ arises because the charge carriers are Cooper pairs ($q = 2e$), not individual electrons. For single-electron systems, $\Phi_0^{(e)} = h/e$.
:::

:::{admonition} Derivation: Flux Quantization from Single-Valuedness
:class: dropdown information

In a superconductor, the Cooper pair condensate is described by a macroscopic order parameter $\Psi(\boldsymbol{r}) = |\Psi|\,\mathrm{e}^{\mathrm{i}\varphi(\boldsymbol{r})}$ with charge $q = 2e$. Inside the bulk of the superconductor, the Meissner effect forces the supercurrent to vanish:

$$

\boldsymbol{j}_s = \frac{q|\Psi|^2}{m}\left(\hbar\nabla\varphi - q\boldsymbol{A}\right) = 0

$$

This gives $\nabla\varphi = q\boldsymbol{A}/\hbar$. Now integrate around a closed contour $\mathcal{C}$ deep inside the superconductor (where $\boldsymbol{j}_s = 0$):

$$

\oint_{\mathcal{C}} \nabla\varphi \cdot \mathrm{d}\boldsymbol{l} = \frac{q}{\hbar}\oint_{\mathcal{C}} \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q\Phi}{\hbar}

$$

The left side must equal $2\pi n$ (integer $n$) because $\Psi$ must be single-valued: $\mathrm{e}^{\mathrm{i}\varphi}$ returns to itself after a full loop. Therefore:

$$

\frac{q\Phi}{\hbar} = 2\pi n \quad \Longrightarrow \quad \Phi = n\frac{2\pi\hbar}{q} = n\frac{h}{2e} = n\Phi_0

$$

where $\Phi_0 = h/(2e) \approx 2.07 \times 10^{-15}\;\text{Wb}$. The derivation has three essential ingredients: (1) the existence of a macroscopic wavefunction, (2) the vanishing of bulk supercurrent (Meissner effect), and (3) single-valuedness of the phase. For a normal metal ring with single electrons ($q = e$), the same argument gives $\Phi_0^{(e)} = h/e$, but thermal fluctuations destroy the phase coherence needed to enforce quantization.
:::

:::{admonition} Derivation: The Factor of 2
:class: dropdown information

In a normal metal, charge carriers are individual electrons ($q = e$), giving $\Phi_0 = h/e$. In a superconductor below $T_c$, electrons condense into Cooper pairs—bound pairs with total charge $2e$ and zero spin. The macroscopic order parameter describes the pair amplitude, so it is the *pair charge* $2e$ that enters the single-valuedness condition. This factor was confirmed experimentally by Doll & Näbauer (1961) and Deaver & Fairbank (1961), providing early evidence for Cooper pairing in BCS theory.
:::

:::{admonition} Discussion
:class: dropdown tip

Flux quantization in a superconductor requires two ingredients: (1) a macroscopic wavefunction with single-valuedness, and (2) Cooper pairing that sets the charge to $2e$. Does knowing the flux quantum $\Phi_0 = h/(2e)$ tell us that pairing *must* occur, or only that paired carriers exist? Could a system have flux quantization with $\Phi_0 = h/e$?
:::

### Experimental Confirmation

**Aharonov-Bohm effect:** Tonomura et al. (1986) used electron holography on tiny solenoids ($\sim$500 nm diameter) to directly observe interference pattern shifts as solenoid current varied, in perfect agreement with theory.

**Flux quantization:** Doll & Näbauer (1961) and Deaver & Fairbank (1961) measured $\Phi_0 = h/(2e)$ in superconducting tin cylinders. Modern SQUIDs routinely exploit flux quantization with extraordinary precision (see [§4.2.3 Flux Ring](4-2-3-flux-ring)).

:::{admonition} Derivation: Topological Nature of the AB Phase — Path Dependence on the Homotopy Class
:class: dropdown information

The Aharonov-Bohm phase is **topological**: it depends only on how many times the path winds around the solenoid, not on the detailed shape of the path. We prove this using the language of homotopy.

**Setup.** The solenoid (radius $R$) carries flux $\Phi_B$. In the exterior region $r > R$, $\boldsymbol{B} = \boldsymbol{0}$ but $\boldsymbol{A} \neq \boldsymbol{0}$ (in e.g.\ the azimuthal gauge $\boldsymbol{A} = \frac{\Phi_B}{2\pi r}\hat{\boldsymbol{\phi}}$).

**Phase along a path.** A particle traveling along path $\mathcal{C}$ acquires phase

$$
\gamma[\mathcal{C}] = \frac{q}{\hbar}\int_{\mathcal{C}} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{r}.
$$

**Deforming the path.** Consider two paths $\mathcal{C}_1$ and $\mathcal{C}_2$ from $P$ to $Q$ with the same endpoints. Their phase difference is

$$
\gamma[\mathcal{C}_1] - \gamma[\mathcal{C}_2] = \frac{q}{\hbar}\oint_{\mathcal{C}_1 - \mathcal{C}_2} \boldsymbol{A}\cdot\mathrm{d}\boldsymbol{r} = \frac{q}{\hbar}\cdot n\Phi_B,
$$

where $n$ is the **winding number** of $\mathcal{C}_1 - \mathcal{C}_2$ around the solenoid. This uses Stokes' theorem on any surface bounded by the loop (which must cross the solenoid interior $n$ times if it winds $n$ times).

**If paths are homotopic** (can be continuously deformed into each other without crossing the solenoid), then $n = 0$ and $\gamma[\mathcal{C}_1] = \gamma[\mathcal{C}_2]$.

**If paths are not homotopic** (one winds around the solenoid once more), $n = 1$ and the phase differs by $q\Phi_B/\hbar$.

**Topological classification.** The fundamental group of the punctured plane $\mathbb{R}^2 \setminus \{\text{solenoid}\}$ is $\pi_1 = \mathbb{Z}$, generated by one loop around the solenoid. Each integer $n$ labels a homotopy class; the AB phase for winding number $n$ is $n \cdot q\Phi_B/\hbar$. This is the sense in which the AB effect is **topological**: it is insensitive to smooth deformations of the path and depends only on global, topological data.
:::

### Summary

- The **Aharonov-Bohm phase** $\Delta\phi = q\Phi_B/\hbar$ arises from the vector potential in field-free regions.
- The **relative phase** between two paths is gauge-invariant and observable through interference.
- The vector potential has **physical reality** in quantum mechanics: gauge-invariant combinations of $\boldsymbol{A}$ are directly measurable.
- **Flux quantization** $\Phi = n\Phi_0$ follows from single-valuedness of the macroscopic wavefunction; $\Phi_0 = h/(2e)$ for Cooper pairs.
- These phenomena are **topological**: they depend only on the winding number of the path, not microscopic details.
