## Lecture Notes

Flux quantization is a profound consequence of quantum mechanics applied to superconductors: the magnetic flux threading a superconducting loop is restricted to discrete values. This is not an approximate effect—it follows directly from the requirement that the macroscopic superconducting wavefunction be single-valued. Flux quantization underlies superconducting quantum devices like SQUIDs, which measure magnetic fields with extraordinary sensitivity, and connects to topological physics.

### Single-Valuedness and Phase Quantization

#### The Single-Valuedness Requirement

A quantum wavefunction must be single-valued: it must return to itself after any closed loop in space. If $\psi(\boldsymbol{r})$ travels around a closed contour $\mathcal{C}$ and returns to its starting point, we require:

$$


\psi(\boldsymbol{r})|_{\text{after loop}} = \psi(\boldsymbol{r})|_{\text{before loop}}

$$


More precisely, the phase accumulated around a closed loop must be a multiple of $2\pi$:

$$


\Delta\phi[\mathcal{C}] = 2\pi n \quad \text{for integer } n

$$


#### Phase from the Aharonov-Bohm Connection

From our discussion of the Aharonov-Bohm effect, the phase accumulated when traversing a closed loop in a region with vector potential $\boldsymbol{A}$ is:

$$


\Delta\phi = \frac{q}{\hbar c}\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l} = \frac{q\Phi_B}{\hbar c}

$$


where $\Phi_B$ is the magnetic flux enclosed by the loop.

### Flux Quantization Condition

For the wavefunction to be single-valued, we require:

$$


\Delta\phi = \frac{q\Phi}{\hbar c} = 2\pi n

$$


Solving for the flux:

$$


\Phi = \frac{2\pi n \hbar c}{q} = n \Phi_0

$$


where we define the **flux quantum**:

$$


\Phi_0 = \frac{2\pi\hbar c}{q}

$$


:::{admonition} Flux Quantization
:class: important

The magnetic flux through a superconducting loop is quantized:

$$


\Phi = n\Phi_0, \quad n = 0, \pm1, \pm2, \ldots

$$


where the flux quantum depends on the charge of the carriers. For a superconductor with Cooper pair condensate (charge $2e$ per pair):

$$


\Phi_{0,\text{Cooper}} = \frac{h}{2e} \approx 2.07 \times 10^{-15} \text{ Wb}

$$


For ordinary electrons (charge $e$):

$$


\Phi_{0,\text{electron}} = \frac{h}\mathrm{e} \approx 4.14 \times 10^{-15} \text{ Wb}

$$


:::

:::{admonition} Discussion
:class: dropdown warning

We derived that $\Delta\phi = 2\pi n$ is required for single-valuedness. But what does "single-valued" really mean physically? The wavefunction $\psi(\boldsymbol{r})$ is already single-valued in the bulk superconductor. Why does the flux through a hole in the ring constrain the phase? What would happen physically if we tried to force $\Phi \neq n\Phi_0$?
:::

### Why Cooper Pairs? The Factor of 2

In a normal metal, charge carriers are individual electrons. But in a superconductor below the critical temperature, electrons condense into **Cooper pairs**—two electrons bound together by the attractive electron-phonon interaction.

Each Cooper pair carries charge $q = 2e$ (two electrons). Therefore, when flux quantization applies to the macroscopic order parameter of the condensate, it is the Cooper pair charge that matters:

$$


\Phi_{0,\text{pairing}} = \frac{h}{2e}

$$


This is the flux quantum that appears in superconductors. The factor of 2 is a fingerprint of the pairing mechanism and is experimentally verifiable.

:::{admonition} Note: Experimental Verification
:class: note

The factor of 2 in superconducting flux quantization was confirmed in a famous 1961 experiment by Doll and Näbauer (and independently by Deaver and Fairbank), using superconducting tin cylinders. This experiment provided direct evidence for Cooper pairing and became a cornerstone of the BCS theory of superconductivity.
:::

:::{admonition} Discussion
:class: dropdown warning

In a normal metal (not superconductor), charge carriers are individual electrons, so the natural flux quantum would be $\Phi_0^{\text{electron}} = h/e$. But electrons don't spontaneously pair in a normal metal—they only pair in a superconductor. How does the superconducting pairing mechanism "know" to produce the factor of 2? Is the flux quantum a consequence of pairing, or does it force pairing to occur?
:::

### Persistent Currents

#### The Meissner Effect

When a superconductor is cooled below its critical temperature, it expels magnetic field from its interior (the Meissner effect). This means the magnetic field in the bulk is zero. But a flux can thread through a "hole" in the superconductor if the material forms a ring or has a hollow cavity.

#### Quantized Persistent Current

If a superconducting ring encloses magnetic flux $\Phi = n\Phi_0$, the superconductor maintains a persistent current to ensure the total enclosed flux remains quantized.

The persistent current is:

$$


I_{\text{persistent}} = \frac{2e}{\hbar}\frac{\mathrm{d}\chi}{\mathrm{d}t}

$$


where $\chi$ is the phase of the macroscopic superconducting order parameter. In equilibrium, $\chi$ is constant (no time evolution), so:

$$


I_{\text{persistent}} = \frac{2e\Phi_{\text{enclosed}}}{\hbar c L}

$$


where $L$ is a self-inductance of the ring. The key point: **this current persists indefinitely** because a superconductor has exactly zero electrical resistance. No energy is dissipated.

:::{admonition} Persistent Currents and Flux Trapping
:class: important

When a superconductor is cooled in an external magnetic field below its critical temperature, the field configuration is "frozen" into the ring (flux trapping). If the field is then removed, the superconductor generates a persistent current that maintains the trapped flux at a quantized value $\Phi = n\Phi_0$.

This trapped flux can be observed by measuring the superconductor's magnetic moment; the flux persists for years with negligible decay.
:::

### SQUIDs: Superconducting Quantum Interference Devices

#### Principle of Operation

A SQUID is a superconducting loop with one or more **Josephson junctions** (thin insulating barriers that allow quantum tunneling of Cooper pairs). The critical current through the SQUID depends sensitively on the enclosed magnetic flux.

:::{admonition} SQUID Critical Current
:class: important

The critical current of a SQUID oscillates periodically with enclosed flux:

$$


I_c(\Phi) = I_c^{\max}\left|\cos\left(\frac{\pi\Phi}{\Phi_0}\right)\right|

$$


The period is exactly one flux quantum $\Phi_0$. This creates an extremely sensitive magnetometer: small changes in flux produce large changes in critical current that are easily measured electronically.
:::

#### Sensitivity

SQUIDs can detect magnetic flux changes of order:

$$


\delta\Phi_{\min} \sim \frac{\Phi_0}{\sqrt{N}}

$$


where $N$ is roughly the number of independent measurements. For a single SQUID, sensitivity approaches $\delta\Phi \sim 10^{-18}$ Wb—orders of magnitude better than conventional magnetometers. This makes SQUIDs the most sensitive magnetic field detectors known.

#:::{admonition} SQUID Applications
:class: example

SQUIDs are used in:
- **Magnetoencephalography (MEG)**: measuring brain magnetic fields (non-invasive neuroimaging)
- **Geomagnetic surveys**: detecting mineral deposits and mapping underground structure
- **Fundamental physics**: searching for magnetic monopoles and axion dark matter
- **Quantum information**: reading out states of superconducting qubits
:::

### Topological Nature of Flux Quantization

Flux quantization is fundamentally topological—it depends only on the winding number of the phase around the loop, not on microscopic details.

:::{admonition} Topology: Why Quantization is Robust
:class: dropdown

Imagine deforming the superconducting ring continuously, or changing its material properties, or altering the electron density. The enclosed flux remains quantized at discrete multiples of $\Phi_0$ *regardless* of these changes.

This robustness arises because flux quantization is a **topological** constraint: it depends only on the topological property that the phase is single-valued, not on the detailed microscopic physics.

This is why SQUIDs work reliably across many different superconducting materials and at different temperatures (as long as $T < T_c$).
:::

### Summary

- **Flux quantization**: Magnetic flux through a superconducting loop must equal $n\Phi_0$ for integer $n$.
- **Flux quantum for Cooper pairs**: $\Phi_0 = h/(2e) \approx 2.07 \times 10^{-15}$ Wb (factor of 2 from pairing).
- **Single-valuedness**: Flux quantization follows from requiring the macroscopic order parameter to be single-valued.
- **Persistent currents**: Superconductors generate persistent currents to maintain quantized flux indefinitely.
- **SQUIDs**: Josephson junction-based SQUIDs leverage flux quantization to create ultra-sensitive magnetometers.
- **Topological robustness**: Flux quantization is independent of microscopic details, making it universal across superconductors.
