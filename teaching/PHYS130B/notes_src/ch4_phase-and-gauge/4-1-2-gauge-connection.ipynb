## Lecture Notes

The gauge connection (vector potential) is the primary object in gauge theory—it is the field that couples directly to quantum particles and governs their dynamics. Unlike the electric and magnetic fields, which are gauge-invariant but derived, the connection is fundamental to quantum mechanics. In this section, we will understand how the connection encodes geometric curvature in phase space, how parallel transport shapes the quantum wavefunction, and why the field strength—not the potential—is what observers actually measure.

### Gauge Connection and Vector Potential

The **gauge connection** $A_i(x, t)$ is the fundamental field that couples to charged quantum particles. In electromagnetism, the connection is identified with the vector potential $\boldsymbol{A}(x, t)$. The gauge connection enters the theory through the **covariant derivative**:

$$

D_i = \partial_i + iqA_i

$$

where $q$ is the charge and $A_i$ is the $i$-th component of the gauge connection (with dimensions of inverse length). Under a local gauge transformation with parameter $\chi(x, t)$:

$$

A_i \to A_i' = A_i - \frac{1}{q}\partial_i\chi

$$

The connection transforms **inhomogeneously**—it doesn't merely multiply by a phase factor. Instead, it shifts by the gradient of the gauge parameter. **Key distinction**: The gauge potential $\boldsymbol{A}$ is not directly observable. Its value depends on the gauge choice. However, certain combinations of $\boldsymbol{A}$ (like line integrals around closed loops) are gauge-invariant and directly observable.

### Parallel Transport and the Connection

The physical meaning of the gauge connection becomes clearer through **parallel transport**. In flat space without a gauge field, moving a wavefunction from position $x$ to a nearby position $x + dx$ would simply be: $\psi(x + dx) = \psi(x) + (\partial_i\psi) dx^i$. However, in the presence of a gauge field, we must adjust the phase by the connection. The **parallel-transported wavefunction** is defined as:

$$

\psi_\parallel(x + dx) = e^{-iqA_i dx^i}\psi(x)

$$

This prescription ensures that the parallel-transported state transforms correctly under gauge transformations. The gauge field therefore measures the **rate of phase change** as we move through space, and the accumulated phase along a path depends on the entire integrated path integral of the connection.

### Field Strength Tensor: Curvature of Gauge Space

The most important gauge-invariant object is the **field strength** or **curvature**, defined as the commutator of covariant derivatives:

$$

F_{ij} = \frac{1}{iq}[D_i, D_j]

$$

This measures how the gauge connection varies from point to point. For **abelian gauge fields** (like electromagnetism), where the gauge fields commute with each other, the field strength has the simple form:

$$

F_{ij} = \partial_i A_j - \partial_j A_i

$$

This is the **field strength tensor** in its fundamental form. It is **antisymmetric**: $F_{ij} = -F_{ji}$. The fact that it arises from the commutator of covariant derivatives shows that field strength is fundamentally about curvature—it measures how parallel transport fails to be commutative when we go around different paths.

:::{admonition} Calculation: Commutator of Covariant Derivatives
:class: dropdown note

Let's compute the commutator carefully. We have:

$$

[D_i, D_j] = (\partial_i + iqA_i)(\partial_j + iqA_j) - (\partial_j + iqA_j)(\partial_i + iqA_i)

$$

Expanding:

$$

[D_i, D_j] = \partial_i\partial_j + iq(\partial_i A_j) + iq A_i\partial_j + (iq)^2 A_i A_j

$$

$$

- [\partial_j\partial_i + iq(\partial_j A_i) + iq A_j\partial_i + (iq)^2 A_j A_i]

$$

The partial derivatives commute: $[\partial_i, \partial_j] = 0$. The $(iq)^2$ terms cancel for abelian (commuting) gauge fields. We're left with:

$$

[D_i, D_j] = iq[(\partial_i A_j) - (\partial_j A_i)] + iq[A_i, A_j]

$$

For **abelian gauge fields** (like electromagnetism), where $[A_i, A_j] = 0$, this simplifies to:

$$

F_{ij} = \frac{1}{iq}[D_i, D_j] = \partial_i A_j - \partial_j A_i

$$

This is the **field strength tensor** in its most fundamental form. It measures how the gauge connection varies from point to point. The field strength is antisymmetric: $F_{ij} = -F_{ji}$, which is why it's naturally a 2-form in differential geometry. In 3D, the 6 independent components of the antisymmetric tensor $F_{ij}$ can be packaged into 3 electric field components and 3 magnetic field components.

:::

### Relation to Electromagnetic Fields

In electromagnetism with 4-vector notation (where index 0 is time), the field strength tensor components are identified with the electric and magnetic fields:

$$

F^{0i} = -F_{0i} = E_i/c \quad (\text{electric field components})

$$

$$

F^{ij} = -\epsilon_{ijk}B_k \quad (\text{magnetic field components})

$$

In 3D spatial notation, with the magnetic field defined as $\boldsymbol{B} = \nabla \times \boldsymbol{A}$ and the electric field as $\boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}$, we recover:

$$

B_k = \epsilon_{ijk}(\partial_i A_j) = \frac{1}{2}\epsilon_{ijk}F_{ij}

$$

The magnetic field is encoded in the spatial components of the field strength tensor, while electric fields come from time variations and scalar potential gradients.

### Gauge Invariance of Field Strength

Under a gauge transformation with parameter $\chi(x, t)$, the vector potential shifts as $A_i \to A_i' = A_i - \frac{1}{q}\partial_i\chi$. What about the field strength? The key observation is that the partial derivatives of the transformation parameter cancel:

$$

F_{ij}' = \partial_i A_j' - \partial_j A_i' = \partial_i\left(A_j - \frac{1}{q}\partial_j\chi\right) - \partial_j\left(A_i - \frac{1}{q}\partial_i\chi\right)

$$

$$

= (\partial_i A_j) - (\partial_j A_i) - \frac{1}{q}[\partial_i\partial_j\chi - \partial_j\partial_i\chi] = F_{ij}

$$

Since partial derivatives commute, **the field strength is completely gauge-invariant!** This is why $F_{ij}$ (or equivalently $\boldsymbol{E}$ and $\boldsymbol{B}$) are the physically observable field quantities—their values are independent of which gauge we choose.

:::{admonition} Proof: Gauge Invariance Calculation
:class: dropdown note

Under a gauge transformation with parameter $\chi(x, t)$:

$$

A_i \to A_i' = A_i - \frac{1}{q}\partial_i\chi

$$

How does the field strength transform?

$$

F_{ij}' = \partial_i A_j' - \partial_j A_i'

$$

$$

= \partial_i\left(A_j - \frac{1}{q}\partial_j\chi\right) - \partial_j\left(A_i - \frac{1}{q}\partial_i\chi\right)

$$

$$

= (\partial_i A_j) - \frac{1}{q}\partial_i\partial_j\chi - (\partial_j A_i) + \frac{1}{q}\partial_j\partial_i\chi

$$

Since partial derivatives commute:

$$

F_{ij}' = (\partial_i A_j) - (\partial_j A_i) = F_{ij}

$$

**Perfect gauge invariance!** The field strength is completely independent of the gauge choice. This is why $F_{ij}$ (or equivalently $\boldsymbol{E}$ and $\boldsymbol{B}$) are the physically observable field quantities.

:::

### Curvature and Parallel Transport Around Closed Loops

The field strength has a profound geometric interpretation through parallel transport. When we transport a wavefunction around a closed loop in space, the phase accumulated depends on the line integral of the connection around that loop. For an infinitesimal rectangular loop in the $ij$-plane with side lengths $dx^i$ and $dx^j$, the net phase accumulated is:

$$

\Delta\phi = qF_{ij}dx^i dx^j

$$

**The field strength measures the phase accumulated when parallel-transporting around infinitesimal loops!** This is not a coincidence—it's a deep geometric principle. For a finite loop, integrating the field strength via Stokes' theorem gives the total phase around the loop. This is the geometric origin of the Aharonov-Bohm effect and shows why even though the vector potential is gauge-dependent, observable physics (phase shifts) depend only on gauge-invariant combinations like the field strength.

:::{admonition} Derivation: Phase Accumulation in Parallel Transport
:class: dropdown note

Consider transporting a wavefunction around an infinitesimal rectangular loop in the $ij$-plane with side lengths $dx^i$ and $dx^j$.

Going around the loop:
- Right: accumulate phase $-qA_i dx^i$
- Up: accumulate phase $-qA_j dx^j$
- Left: accumulate phase $+qA_i(x + dx^j) dx^i \approx -qA_i dx^i + qA_i(\partial_j dx^j) dx^i$
- Down: accumulate phase $+qA_j(x + dx^i) dx^j \approx -qA_j dx^j + qA_j(\partial_i dx^i) dx^j$

The net accumulated phase is:

$$

\Delta\phi = q(\partial_i A_j)dx^i dx^j - q(\partial_j A_i)dx^i dx^j = qF_{ij}dx^i dx^j

$$

**The field strength measures the phase accumulated when parallel-transporting around infinitesimal loops!** For a finite loop, integrating the field strength (via Stokes' theorem) gives the total phase around the loop. This is the geometric origin of the Aharonov-Bohm effect.

:::

### Non-Abelian Gauge Fields

For non-commuting gauge fields (like in Yang-Mills theory or spin systems), the field strength is:

$$

F_{ij} = \partial_i A_j - \partial_j A_i + \frac{1}{iq}[A_i, A_j]

$$

The commutator term $[A_i, A_j]$ appears when the gauge fields are matrices (non-abelian). This makes the theory much richer but also more complex. The interaction between field components leads to phenomena like color confinement in QCD. For this course, we focus on abelian $U(1)$ gauge theory (electromagnetism and magnetic monopoles), where this extra term vanishes and field strength reduces to the simple form $F_{ij} = \partial_i A_j - \partial_j A_i$.

### Summary

- **Gauge connection**: The field $A_i(x, t)$ couples charged particles to gauge interactions through the covariant derivative.
- **Gauge potential**: $\boldsymbol{A}$ is gauge-dependent (unobservable) but fundamental to quantum dynamics.
- **Parallel transport**: Moving a wavefunction through space requires phase adjustment by the connection: $\psi_\parallel(x + dx) = e^{-iqA_i dx^i}\psi(x)$.
- **Field strength**: $F_{ij} = \partial_i A_j - \partial_j A_i$ (for abelian fields) measures curvature of gauge space.
- **Gauge invariance**: Field strength is completely gauge-invariant, unlike the potential.
- **Geometric interpretation**: Field strength encodes the phase accumulated when parallel-transporting around closed loops.
- **Connection to EM fields**: Spatial components of $F_{ij}$ give magnetic field $\boldsymbol{B}$; time components give electric field $\boldsymbol{E}$.
- **Curvature principle**: The commutator of covariant derivatives directly yields the field strength, illustrating the deep connection to differential geometry.
