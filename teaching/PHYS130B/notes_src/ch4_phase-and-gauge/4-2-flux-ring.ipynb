# 4.2 Flux Ring

### Overview

The Aharonov-Bohm effect demonstrates that vector potentials have direct physical significance in quantum mechanics—they modify quantum phases even in field-free regions. Magnetic flux threading topological loops is quantized, a fundamental constraint on quantum systems.

The Aharonov-Bohm effect proves that **electromagnetic potentials are fundamental in quantum mechanics**, not just auxiliary fields. It demonstrates quantum non-locality: particles couple to potentials at distant locations even when fields vanish everywhere along their paths.

Flux quantization explains charge quantization (Dirac's condition) and underlies superconducting devices like SQUIDs that measure magnetic fields to extraordinary precision.

:::{admonition} Topological vs Geometric
:class: attention

Flux quantization is a **topological** effect: it depends only on how many times the phase winds around a loop, not on microscopic details. This topological robustness is why superconducting devices are so reliable.
:::
