## Lecture Notes

### Orbital Angular Momentum

The orbital angular momentum operator is defined as:

$$ \hat{\boldsymbol{L}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{p}}$$


In Cartesian components:

$$ \hat{L}_i = \epsilon_{ijk}\hat{r}_j\hat{p}_k$$


These operators satisfy the canonical **angular momentum commutation relations**:

$$ [\hat{L}_i, \hat{L}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{L}_k$$


In particular, $[\hat{L}_x, \hat{L}_y] = \mathrm{i}\hbar\hat{L}_z$ (and cyclic permutations). The magnitude squared commutes with all components:

$$ [\hat{L}^2, \hat{L}_i] = 0 \quad \text{for all } i$$


where $\hat{L}^2 = \hat{L}_x^2 + \hat{L}_y^2 + \hat{L}_z^2$.

:::{admonition} Key Property: Orbital $L$ Quantization
:class: important

Because orbital angular momentum arises from a spatial degree of freedom, its eigenvalues are **integers only**:

$$ \hat{L}^2 Y_l^m(\theta,\phi) = \hbar^2 l(l+1) Y_l^m(\theta,\phi), \quad l = 0, 1, 2, \ldots$$


$$ \hat{L}_z Y_l^m(\theta,\phi) = \hbar m Y_l^m(\theta,\phi), \quad m = -l, \ldots, l$$


The eigenfunctions are the **spherical harmonics** $Y_l^m(\theta,\phi)$. The quantum numbers $l$ and $m$ are always integers.
:::

### Intrinsic Spin Angular Momentum

Unlike orbital angular momentum, **spin** $\hat{\boldsymbol{S}}$ is an intrinsic quantum property—it does not arise from spatial motion. It is an internal angular momentum that has no classical analogue.

For a spin-1/2 particle (like an electron or proton):

$$ \hat{\boldsymbol{S}} = \frac{\hbar}{2}\hat{\boldsymbol{\sigma}}$$


where $\hat{\boldsymbol{\sigma}} = (\hat{\sigma}^x, \hat{\sigma}^y, \hat{\sigma}^z)$ are the **Pauli matrices**:

$$ \hat{\sigma}^x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}, \quad \hat{\sigma}^y = \begin{pmatrix} 0 & -\mathrm{i} \\ \mathrm{i} & 0 \end{pmatrix}, \quad \hat{\sigma}^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$$


The spin operators satisfy the same commutation relations as orbital angular momentum:

$$ [\hat{S}_i, \hat{S}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{S}_k$$


However, the **eigenvalues** differ. For spin-1/2:

$$ \hat{S}^2|\chi\rangle = \frac{3}{4}\hbar^2|\chi\rangle \quad \text{(for all states)}$$


$$ \hat{S}_z|\uparrow\rangle = \frac{\hbar}{2}|\uparrow\rangle, \quad \hat{S}_z|\downarrow\rangle = -\frac{\hbar}{2}|\downarrow\rangle$$


The **spin quantum numbers are half-integer**: $s = 1/2$, and $m_s = \pm 1/2$.

:::{admonition} Key Property: Half-Integer Spin Exists
:class: important

Unlike orbital angular momentum, spin can have **half-integer quantum numbers**. A spin-1/2 particle is described by a 2-dimensional Hilbert space (a spinor):

$$ |\chi\rangle = \begin{pmatrix} \chi_\uparrow \\ \chi_\downarrow \end{pmatrix}$$


where $|\uparrow\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$ and $|\downarrow\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.

This is fundamentally different from $L$, which always produces integers.
:::

### Total Angular Momentum

When both orbital and spin angular momentum are present, the **total angular momentum** is:

$$ \hat{\boldsymbol{J}} = \hat{\boldsymbol{L}} + \hat{\boldsymbol{S}}$$


The total $J$ operator satisfies the same commutation relations:

$$ [\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$$


By the angular momentum addition rules, if a system has orbital quantum number $l$ and spin quantum number $s$, then:

$$ j = |l - s|, |l - s| + 1, \ldots, l + s$$


For example, an electron in a $p$-orbital ($l = 1$) with spin-1/2 ($s = 1/2$) can have $j = 1/2$ or $j = 3/2$.

The quantum numbers $(l, s, j, m_j)$ are useful when $\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}$ coupling (spin-orbit coupling) is significant.

### Spin-Orbit Coupling

When a charged particle moves in a potential with a radial force (such as a nucleus), **relativistic effects** lead to a coupling between spin and orbital angular momentum.

:::{admonition} Derivation: Spin-Orbit Coupling Energy
:class: dropdown

In the rest frame of the electron orbiting a nucleus with charge $Ze$, the nucleus appears to move, creating an effective magnetic field. This field couples to the electron's magnetic moment:

$$ \boldsymbol{B}_\text{eff} = -\frac{1}{c^2m}(\boldsymbol{E} \times \boldsymbol{v})$$


For a central potential $V(r) = -\frac{Ze^2}{r}$, the electric field is $\boldsymbol{E} = -\nabla V$, so:

$$ \boldsymbol{E} \times \boldsymbol{v} = -\frac{1}{r}\frac{\mathrm{d}V}{\mathrm{d}r}(\hat{r} \times \boldsymbol{v})$$


The electron's magnetic moment is $\boldsymbol{\mu} = -\frac{g e}{2m c}\hat{\boldsymbol{S}}$ (with $g \approx 2$ for the electron), so the energy is:

$$ H_{SO} = \boldsymbol{\mu} \cdot \boldsymbol{B}_\text{eff}$$


After accounting for **Thomas precession** (a relativistic geometric effect that introduces a factor of 1/2), one obtains:

$$ H_{SO} = \frac{1}{2m^2c^2r}\frac{\mathrm{d}V}{\mathrm{d}r}\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}$$


For the Coulomb potential, $\frac{\mathrm{d}V}{\mathrm{d}r} = \frac{Ze^2}{r^2}$, giving:

$$ H_{SO} = \frac{Ze^2}{2m^2c^2r^3}\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}$$


:::

The strength of spin-orbit coupling is parametrized as:

$$ H_{SO} = \xi(r)\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}$$


where $\xi(r) = \frac{1}{2m^2c^2r}\frac{\mathrm{d}V}{\mathrm{d}r}$ depends on the potential. For hydrogen-like atoms, $\xi(r) \propto 1/r^3$.

:::{admonition} Key Formula: Dot Product via Total $J$
:class: important

The dot product $\hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}}$ can be written in terms of the total angular momentum:

$$ \hat{\boldsymbol{L}} \cdot \hat{\boldsymbol{S}} = \frac{1}{2}(\hat{J}^2 - \hat{L}^2 - \hat{S}^2)$$


Since $\hat{J}^2$, $\hat{L}^2$, and $\hat{S}^2$ are diagonal in the $|l, s, j, m_j\rangle$ basis, this formula greatly simplifies perturbation theory calculations.

For a given $l$, $s$, and $j$:

$$ \langle \hat{L} \cdot \hat{S} \rangle = \frac{\hbar^2}{2}[j(j+1) - l(l+1) - s(s+1)]$$

:::

**Physical consequence**: Spin-orbit coupling splits energy levels. For example, in hydrogen's $2p$ state:
- Without coupling: $2p$ is 3-fold degenerate (three $m_l$ values with any spin)
- With coupling: the state splits into $2p_{3/2}$ ($j=3/2$, 4-fold degenerate) and $2p_{1/2}$ ($j=1/2$, 2-fold degenerate)

This **fine structure splitting** was a key piece of evidence that spin-orbit coupling is real.

:::{admonition} Discussion: Spin-Orbit Coupling
:class: dropdown warning

Spin-orbit coupling ∝ L·S mixes eigenstates of L_z and S_z.

:::

### Representations of the Rotation Group

A fundamental distinction between orbital and spin angular momentum emerges from group theory:

- **Orbital angular momentum** generates **integer** representations of $SO(3)$ (the rotation group in 3D)
- **Spin** generates both integer and **half-integer** representations of $SU(2)$ (the double cover of $SO(3)$)

This distinction is topological. Under a rotation by $2\pi$ around any axis:
- A state with integer $l$ returns to itself: $|\psi\rangle \to |\psi\rangle$
- A state with half-integer $s$ picks up a minus sign: $|\chi\rangle \to -|\chi\rangle$

This is observable only for half-integer spin; under *two* consecutive $2\pi$ rotations, any state returns to itself. This is why interference experiments can distinguish half-integer from integer angular momentum.

### Summary

| Property | Orbital $\hat{\boldsymbol{L}}$ | Spin $\hat{\boldsymbol{S}}$ |
|----------|-----------|-----------|
| **Origin** | Spatial motion | Intrinsic to particle |
| **Quantum numbers** | $l = 0, 1, 2, \ldots$ (integer only) | $s = 1/2, 1, 3/2, \ldots$ (integer or half-integer) |
| **For $s=1/2$** | N/A | $m_s = \pm 1/2$ |
| **Eigenvalues of $L^2$** | $\hbar^2 l(l+1)$ | $\hbar^2 s(s+1)$ |
| **Representation of $SO(3)$** | Integer reps only | Integer + half-integer reps |
| **Interaction with field** | Couples to electric fields | Couples to magnetic fields |
| **Commutation relations** | $[L_i, L_j] = \mathrm{i}\hbar\epsilon_{ijk}L_k$ | $[S_i, S_j] = \mathrm{i}\hbar\epsilon_{ijk}S_k$ |
