## Lecture Notes

Coupling a charged particle to an electromagnetic field is the next step in understanding gauge theory. We move from the abstract principle of gauge invariance to concrete physical consequences: how charges respond to fields, how energy and momentum are redefined, and how the kinetic momentum emerges as the physically relevant quantity.

### Minimal Coupling Prescription

#### From Classical to Quantum

In classical mechanics, a charged particle in an electromagnetic field obeys the Lorentz force:

$$


m\frac{\mathrm{d}^2\boldsymbol{r}}{\mathrm{d}t^2} = q(\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B})

$$


The Hamiltonian formulation expresses this through the canonical momentum:

$$


H_{\text{cl}} = \frac{(\boldsymbol{p} - q\boldsymbol{A})^2}{2m} + q\Phi

$$


where $\boldsymbol{p}$ is the canonical momentum (different from $m\boldsymbol{v}$ when fields are present), and $\boldsymbol{A}$, $\Phi$ are the vector and scalar potentials.

To quantize a charged particle in an electromagnetic field, we use the **minimal coupling prescription**:

:::{admonition} Minimal Coupling Prescription
:class: important

Replace the canonical momentum operator with:

$$


\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}(\hat{\boldsymbol{r}})

$$


The quantum Hamiltonian becomes:

$$


\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi

$$


This is equivalent to the gauge-covariant Schrödinger equation where ordinary derivatives are replaced by covariant derivatives $D_i = \partial_i + \mathrm{i}qA_i$.
:::

This prescription is not arbitrary—it is the unique choice that preserves gauge covariance of the Schrödinger equation.

#### Expansion of the Hamiltonian

Expand the kinetic energy term:

$$


(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 = \hat{\boldsymbol{p}}^2 - q(\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}}) + q^2\boldsymbol{A}^2

$$


Since $\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla$:

$$


\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}} = -\mathrm{i}\hbar(\nabla \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \nabla + \nabla \cdot \boldsymbol{A})

$$


In the **Coulomb gauge** where $\nabla \cdot \boldsymbol{A} = 0$:

$$


\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}} = -2\mathrm{i}\hbar\boldsymbol{A} \cdot \nabla = 2\hat{\boldsymbol{p}} \cdot \boldsymbol{A}

$$


:::{admonition} Hamiltonian in Coulomb Gauge
:class: important

In the Coulomb gauge ($\nabla \cdot \boldsymbol{A} = 0$):

$$


\hat{H} = \frac{\hat{\boldsymbol{p}}^2}{2m} - \frac{q}{m}\boldsymbol{A} \cdot \hat{\boldsymbol{p}} + \frac{q^2}{2m}\boldsymbol{A}^2 + q\Phi

$$


- **First term**: kinetic energy (field-free)
- **Second term**: linear coupling to magnetic field (diamagnetic coupling)
- **Third term**: quadratic response to field (important only for strong fields)
- **Fourth term**: Coulomb potential from scalar field
:::

### Canonical Momentum vs. Kinetic Momentum

A crucial distinction in quantum mechanics with electromagnetic fields is between two types of momentum:

:::{admonition} Canonical vs. Kinetic Momentum
:class: important

The **canonical momentum** is:

$$


\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla

$$


It satisfies the canonical commutation relation:

$$


[\hat{p}_i, \hat{x}_j] = -\mathrm{i}\hbar\delta_{ij}

$$


The **kinetic momentum** (or physical momentum) is:

$$


\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}

$$


It determines the particle's actual velocity and obeys:

$$


[\hat{\pi}_i, \hat{x}_j] = -\mathrm{i}\hbar\delta_{ij} - q\partial_j A_i

$$


The correction term depends on how the magnetic field varies in space.
:::

Why the distinction? The **canonical momentum commutes with the Hamiltonian** (in certain gauges), making it a useful quantum number. But the **kinetic momentum** represents the actual velocity and appears in physical observables like current and energy flux.

### Lorentz Force from Quantum Mechanics

#### Heisenberg Equation of Motion

Use the Heisenberg equation to derive the force law from the Hamiltonian:

$$


\frac{\mathrm{d}\hat{\boldsymbol{p}}}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}[\hat{\boldsymbol{p}}, \hat{H}]

$$


In the Coulomb gauge with time-dependent fields, the calculation yields:

$$


m\frac{\mathrm{d}\langle\boldsymbol{v}\rangle}{\mathrm{d}t} = q\langle\boldsymbol{E}\rangle + q\langle\boldsymbol{v} \times \boldsymbol{B}\rangle

$$


:::{admonition} Derivation: Quantum Lorentz Force
:class: dropdown

Start with:

$$


\frac{\mathrm{d}\hat{\boldsymbol{p}}}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}[\hat{\boldsymbol{p}}, \hat{H}] = \frac{1}{\mathrm{i}\hbar}[\hat{\boldsymbol{p}}, -\frac{q}{m}\boldsymbol{A} \cdot \hat{\boldsymbol{p}} + q\Phi + \ldots]

$$


The commutator $[\hat{p}_i, A_j\hat{p}_j]$ requires the product rule:

$$


[\hat{p}_i, A_j\hat{p}_j] = [\hat{p}_i, A_j]\hat{p}_j + A_j[\hat{p}_i, \hat{p}_j]

$$


Since $[\hat{p}_i, \hat{p}_j] = 0$ and $[\hat{p}_i, A_j] = -\mathrm{i}\hbar\partial_i A_j$:

$$


[\hat{p}_i, A_j\hat{p}_j] = -\mathrm{i}\hbar(\partial_i A_j)\hat{p}_j

$$


Similarly, $[\hat{p}_i, \Phi] = -\mathrm{i}\hbar\partial_i\Phi$.

Collecting all terms:

$$


\frac{\mathrm{d}\langle\hat{\boldsymbol{p}}\rangle}{\mathrm{d}t} = q\langle\nabla_{\boldsymbol{r}}\Phi\rangle - \frac{q}{m}\langle(\nabla_{\boldsymbol{r}} A_i)\hat{p}_i\rangle + q\frac{\partial\langle\boldsymbol{A}\rangle}{\partial t}

$$


Using $\boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}$ and $\boldsymbol{B} = \nabla \times \boldsymbol{A}$ (which relates $(\partial_i A_j - \partial_j A_i)$ to $\boldsymbol{B}$):

$$


m\frac{\mathrm{d}\langle\boldsymbol{v}\rangle}{\mathrm{d}t} = q\langle\boldsymbol{E}\rangle + q\langle\boldsymbol{v} \times \boldsymbol{B}\rangle

$$


:::

This is the **correspondence principle in action**: expectation values of quantum observables obey classical equations of motion.

:::{admonition} Discussion
:class: dropdown warning

The Coulomb gauge ($\nabla \cdot \boldsymbol{A} = 0$) is convenient for atomic physics calculations. However, in a region where $\nabla \cdot \boldsymbol{A} \neq 0$, that gauge choice is not available. Does this mean different physical regimes require different gauges? Or should all physical predictions be independent of which gauge we choose?
:::

### Gauge Invariance of the Schrödinger Equation

The Schrödinger equation with electromagnetic fields is manifestly gauge-covariant:

:::{admonition} Gauge Covariance of the Schrödinger Equation
:class: important

If $\psi(\boldsymbol{x}, t)$ satisfies:

$$


\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = \hat{H}[\boldsymbol{A}, \Phi]\psi

$$


then $\psi'(\boldsymbol{x}, t) = \mathrm{e}^{\mathrm{i}q\chi(\boldsymbol{x}, t)}\psi(\boldsymbol{x}, t)$ satisfies:

$$


\mathrm{i}\hbar\frac{\partial\psi'}{\partial t} = \hat{H}[\boldsymbol{A}', \Phi']\psi'

$$


where $\boldsymbol{A}' = \boldsymbol{A} - \nabla\chi$ and $\Phi' = \Phi + \frac{1}{q}\partial_t\chi$.

This means: **different gauge choices yield wavefunctions that differ by a phase, but produce identical predictions for all physical observables**.
:::

### Probability Current with Electromagnetic Field

A fundamental consequence of gauge invariance is how probability current transforms. The continuity equation is:

$$


\frac{\partial\rho}{\partial t} + \nabla \cdot \boldsymbol{J} = 0

$$


where $\rho = |\psi|^2$ is probability density. With an electromagnetic field, the probability current is:

:::{admonition} Probability Current in Electromagnetic Field
:class: important

$$


\boldsymbol{J} = \frac{1}{m}\text{Im}(\psi^*\hat{\boldsymbol{\pi}}\psi) = \frac{1}{m}\text{Im}(\psi^*[\hat{\boldsymbol{p}} - q\boldsymbol{A}]\psi)

$$


The **kinetic momentum** $\hat{\boldsymbol{\pi}}$ appears naturally in the current, not the canonical momentum. This makes physical sense: the current describes actual particle flow, which is determined by kinetic velocity $\boldsymbol{v} = \hat{\boldsymbol{\pi}}/m$, not canonical velocity.
:::

### Summary

- The **minimal coupling prescription** $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$ is the unique choice preserving gauge covariance.
- The **kinetic momentum** $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ represents actual particle momentum; canonical momentum $\hat{\boldsymbol{p}}$ is a mathematical tool.
- Expectation values obey the **quantum Lorentz force**, confirming the correspondence principle.
- The Schrödinger equation is **gauge-covariant**: different gauges give phase-related wavefunctions with identical physics.
- **Probability current** naturally involves kinetic momentum, reflecting the physical meaning of particle flow in electromagnetic fields.
