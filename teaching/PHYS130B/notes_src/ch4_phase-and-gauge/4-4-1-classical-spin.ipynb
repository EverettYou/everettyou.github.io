## Lecture Notes

### Overview

The classical spinning top provides the physical intuition for angular momentum: a rigid body possesses both orbital angular momentum $\boldsymbol{L} = \boldsymbol{r} \times \boldsymbol{p}$ from translational motion and spin angular momentum $\boldsymbol{S}$ from rotation about its own axis. A spinning charge generates a magnetic dipole, coupling angular momentum to electromagnetic fields through the gyromagnetic ratio. Precession in a magnetic field — Larmor precession — is the classical prototype for quantum spin dynamics. Yet a deep puzzle remains: orbital angular momentum on a sphere is always integer-valued, while experiments (Stern-Gerlach) reveal half-integer angular momentum. Where does spin-1/2 come from? The answer, developed in §4.4.2–4.4.3, lies in gauge topology.

### The Classical Spinning Top

A rigid body rotating about its center of mass has angular momentum:

$$
\boldsymbol{L} = I \boldsymbol{\omega}
$$

where $I$ is the moment of inertia tensor and $\boldsymbol{\omega}$ is the angular velocity. For a symmetric top with symmetry axis $\hat{\boldsymbol{n}}$, the angular momentum decomposes into two parts:

- **Orbital part** $\boldsymbol{L}_\text{orb} = \boldsymbol{r} \times \boldsymbol{p}$: angular momentum from the center-of-mass motion
- **Spin part** $\boldsymbol{S}$: angular momentum from rotation about the body's own axis

The total angular momentum is $\boldsymbol{J} = \boldsymbol{L}_\text{orb} + \boldsymbol{S}$.

:::{admonition} Orbital vs Spin Angular Momentum
:class: important

**Orbital angular momentum** $\boldsymbol{L} = \boldsymbol{r} \times \boldsymbol{p}$ arises from motion through space. It depends on the choice of origin.

**Spin angular momentum** $\boldsymbol{S}$ is intrinsic rotation about the body's own axis. It is independent of the choice of origin.

Both obey the same dynamical equation: $\mathrm{d}\boldsymbol{J}/\mathrm{d}t = \boldsymbol{\tau}$, where $\boldsymbol{\tau}$ is the applied torque.
:::

For a symmetric top with moment of inertia $I_\parallel$ about the symmetry axis and $I_\perp$ about any transverse axis, the kinetic energy is:

$$
T = \frac{1}{2} I_\perp \omega_\perp^2 + \frac{1}{2} I_\parallel \omega_\parallel^2
$$

The spin component $S = I_\parallel \omega_\parallel$ along the symmetry axis is conserved when no torque acts along that axis.

:::{admonition} Derivation: Equation of Motion for the Spinning Axis
:class: dropdown information

Starting from $\boldsymbol{\tau} = \mathrm{d}\boldsymbol{L}/\mathrm{d}t$ with $\boldsymbol{L} = I_\perp \boldsymbol{\Omega} + I_\parallel \boldsymbol{\omega}$:

$$
\boldsymbol{\tau} = I_\parallel \frac{\mathrm{d}\boldsymbol{\omega}}{\mathrm{d}t} + I_\perp \frac{\mathrm{d}\boldsymbol{\Omega}}{\mathrm{d}t}
$$

**Step 1.** Under the transverse-torque assumption, only the *direction* of $\boldsymbol{\omega} = \omega \hat{\boldsymbol{n}}$ changes, so $\mathrm{d}\boldsymbol{\omega}/\mathrm{d}t = \boldsymbol{\Omega} \times \boldsymbol{\omega}$.

**Step 2.** The unit vector $\hat{\boldsymbol{n}}$ rotates as $\dot{\hat{\boldsymbol{n}}} = \boldsymbol{\Omega} \times \hat{\boldsymbol{n}}$. Cross-multiplying both sides from the left by $\hat{\boldsymbol{n}}$:

$$
\hat{\boldsymbol{n}} \times \dot{\hat{\boldsymbol{n}}} = \hat{\boldsymbol{n}} \times (\boldsymbol{\Omega} \times \hat{\boldsymbol{n}}) = \boldsymbol{\Omega}(\hat{\boldsymbol{n}} \cdot \hat{\boldsymbol{n}}) - \hat{\boldsymbol{n}}(\hat{\boldsymbol{n}} \cdot \boldsymbol{\Omega}) = \boldsymbol{\Omega}
$$

since $\hat{\boldsymbol{n}} \cdot \hat{\boldsymbol{n}} = 1$ and $\hat{\boldsymbol{n}} \cdot \boldsymbol{\Omega} = 0$ ($\boldsymbol{\Omega}$ is transverse).

**Step 3.** Substitute $\boldsymbol{\Omega} = \hat{\boldsymbol{n}} \times \dot{\hat{\boldsymbol{n}}}$ into the torque equation:

$$
\boldsymbol{\tau} = I_\parallel (\hat{\boldsymbol{n}} \times \dot{\hat{\boldsymbol{n}}}) \times \boldsymbol{\omega} + I_\perp \frac{\mathrm{d}}{\mathrm{d}t}(\hat{\boldsymbol{n}} \times \dot{\hat{\boldsymbol{n}}})
$$

Expand the first term using the BAC-CAB rule: $(\hat{\boldsymbol{n}} \times \dot{\hat{\boldsymbol{n}}}) \times \boldsymbol{\omega} = (\hat{\boldsymbol{n}} \cdot \boldsymbol{\omega})\dot{\hat{\boldsymbol{n}}} - \hat{\boldsymbol{n}}(\dot{\hat{\boldsymbol{n}}} \cdot \boldsymbol{\omega})$. Since $\dot{\hat{\boldsymbol{n}}} \cdot \boldsymbol{\omega} = \omega(\dot{\hat{\boldsymbol{n}}} \cdot \hat{\boldsymbol{n}}) = 0$ (from $\mathrm{d}(\hat{\boldsymbol{n}} \cdot \hat{\boldsymbol{n}})/\mathrm{d}t = 0$), this simplifies to $\omega \dot{\hat{\boldsymbol{n}}}$.

For the second term: $\frac{\mathrm{d}}{\mathrm{d}t}(\hat{\boldsymbol{n}} \times \dot{\hat{\boldsymbol{n}}}) = \dot{\hat{\boldsymbol{n}}} \times \dot{\hat{\boldsymbol{n}}} + \hat{\boldsymbol{n}} \times \ddot{\hat{\boldsymbol{n}}} = \hat{\boldsymbol{n}} \times \ddot{\hat{\boldsymbol{n}}}$.

**Step 4.** With $S = I_\parallel \omega$, the equation becomes:

$$
\boldsymbol{\tau} = S\dot{\hat{\boldsymbol{n}}} + I_\perp \hat{\boldsymbol{n}} \times \ddot{\hat{\boldsymbol{n}}}
$$

Cross-multiplying by $\hat{\boldsymbol{n}}$ from the right:

$$
\boldsymbol{\tau} \times \hat{\boldsymbol{n}} = S\dot{\hat{\boldsymbol{n}}} \times \hat{\boldsymbol{n}} + I_\perp (\hat{\boldsymbol{n}} \times \ddot{\hat{\boldsymbol{n}}}) \times \hat{\boldsymbol{n}}
$$

The last factor simplifies via BAC-CAB: $(\hat{\boldsymbol{n}} \times \ddot{\hat{\boldsymbol{n}}}) \times \hat{\boldsymbol{n}} = (\hat{\boldsymbol{n}} \cdot \hat{\boldsymbol{n}})\ddot{\hat{\boldsymbol{n}}} - \hat{\boldsymbol{n}}(\ddot{\hat{\boldsymbol{n}}} \cdot \hat{\boldsymbol{n}}) = (\ddot{\hat{\boldsymbol{n}}})_\perp$, which is the tangential component of the acceleration. Therefore:

$$
I_\perp (\ddot{\hat{\boldsymbol{n}}})_\perp = \boldsymbol{\tau} \times \hat{\boldsymbol{n}} - \dot{\hat{\boldsymbol{n}}} \times \boldsymbol{S}
$$

This is Newton's second law on the sphere: the tangential acceleration equals the torque-induced force minus a velocity-dependent "magnetic" force from the spin.
:::

### Precession in a Torque Field

When a torque $\boldsymbol{\tau}$ acts on a spinning body, the angular momentum evolves as:

$$
\frac{\mathrm{d}\boldsymbol{L}}{\mathrm{d}t} = \boldsymbol{\tau}
$$

For a magnetic dipole $\boldsymbol{\mu}$ in a uniform field $\boldsymbol{B}$, the torque is $\boldsymbol{\tau} = \boldsymbol{\mu} \times \boldsymbol{B}$. If the magnetic moment is proportional to the angular momentum, $\boldsymbol{\mu} = \gamma \boldsymbol{L}$, then:

$$
\frac{\mathrm{d}\boldsymbol{L}}{\mathrm{d}t} = \gamma \boldsymbol{L} \times \boldsymbol{B}
$$ (eq-classical-spin-precession)

This equation describes **precession**: the angular momentum vector traces a cone around $\boldsymbol{B}$ at a constant angle $\theta$, with angular frequency:

$$
\omega_L = -\gamma B
$$ (eq-larmor-frequency)

This is the **Larmor frequency**. The magnitude $|\boldsymbol{L}|$ and the component $L_z = \boldsymbol{L} \cdot \hat{\boldsymbol{B}}$ are both conserved — only the transverse components rotate.

:::{admonition} Derivation: Poisson Brackets for Angular Momentum and the Precession Equation
:class: dropdown information

We show that the precession equation $\dot{\boldsymbol{L}} = \gamma\boldsymbol{L}\times\boldsymbol{B}$ follows directly from Hamilton's equations using the Poisson bracket algebra of angular momentum — and that quantizing these brackets yields the quantum commutation relations.

**Step 1: Compute $\{L_x, L_y\}$ from canonical Poisson brackets.** For $L_x = yp_z - zp_y$ and $L_y = zp_x - xp_z$ with canonical brackets $\{q_i, p_j\} = \delta_{ij}$, $\{q_i, q_j\} = \{p_i, p_j\} = 0$:

$$

\{L_x, L_y\} = \sum_i\left(\frac{\partial L_x}{\partial q_i}\frac{\partial L_y}{\partial p_i} - \frac{\partial L_x}{\partial p_i}\frac{\partial L_y}{\partial q_i}\right)

$$

Only the $z$-terms contribute (the $x$ and $y$ terms vanish):

- $i = x$: $(\partial_x L_x)(\partial_{p_x} L_y) - (\partial_{p_x} L_x)(\partial_x L_y) = (0)(z) - (0)(-p_z) = 0$
- $i = y$: $(p_z)(0) - (-z)(0) = 0$
- $i = z$: $(-p_y)(-x) - (y)(p_x) = xp_y - yp_x = L_z$

So $\{L_x, L_y\} = L_z$. By cyclic permutation: $\boldsymbol{\{L_i, L_j\} = \epsilon_{ijk}L_k}$.

**Step 2: Hamilton's equations give the precession.** For a magnetic dipole $\boldsymbol{\mu} = \gamma\boldsymbol{L}$ in field $\boldsymbol{B} = B\hat{\boldsymbol{z}}$, the Hamiltonian is $H = -\gamma B L_z$. Hamilton's equations $\dot{L}_i = \{L_i, H\}$:

$$

\dot{L}_x = \{L_x,\,-\gamma B L_z\} = -\gamma B\{L_x, L_z\} = -\gamma B(-L_y) = \gamma B L_y

$$

$$

\dot{L}_y = \{L_y,\,-\gamma B L_z\} = -\gamma B\{L_y, L_z\} = -\gamma B L_x

$$

where we used $\{L_y, L_z\} = \epsilon_{yzk}L_k = L_x$ (cyclic from $\{L_x,L_y\}=L_z$).

And $\dot{L}_z = \{L_z, -\gamma BL_z\} = 0$. Together:

$$

\dot{\boldsymbol{L}} = \gamma B(L_y\hat{x} - L_x\hat{y}) = \gamma\boldsymbol{L}\times\boldsymbol{B}

$$

This is exactly equation {eq}`eq-classical-spin-precession`, now derived from Hamilton's equations rather than postulated.

**Step 3: Canonical quantization.** The correspondence principle replaces $\{A,B\} \to [\hat{A},\hat{B}]/(\mathrm{i}\hbar)$. Applied to $\{L_i, L_j\} = \epsilon_{ijk}L_k$:

$$

[\hat{L}_i, \hat{L}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{L}_k

$$

This is the quantum angular momentum algebra derived purely from $\hat{\boldsymbol{L}} = \hat{\boldsymbol{r}}\times\hat{\boldsymbol{p}}$. The same algebra holds for intrinsic spin $\hat{\boldsymbol{S}}$ — even though spin has no classical orbital interpretation — which is why $j = 1/2$ representations are allowed by the algebra even if they cannot come from orbital motion.
:::

:::{admonition} Larmor Precession
:class: important

A magnetic dipole with $\boldsymbol{\mu} = \gamma \boldsymbol{L}$ in a field $\boldsymbol{B} = B \hat{\boldsymbol{z}}$ precesses at the **Larmor frequency** $\omega_L = -\gamma B$. The equation of motion,

$$
\frac{\mathrm{d}\boldsymbol{L}}{\mathrm{d}t} = \gamma \boldsymbol{L} \times \boldsymbol{B}
$$

is formally identical to the Lorentz force on a charged particle, with $\boldsymbol{L}$ playing the role of velocity and $\boldsymbol{B}$ playing the role of the magnetic field.
:::

:::{admonition} Derivation: Solving the Precession Equation
:class: dropdown information

For $\boldsymbol{B} = B\hat{\boldsymbol{z}}$, the component equations are:

$$
\dot{L}_x = \gamma B L_y, \quad \dot{L}_y = -\gamma B L_x, \quad \dot{L}_z = 0
$$

Define $L_\pm = L_x \pm \mathrm{i} L_y$. Then $\dot{L}_\pm = \mp \mathrm{i} \gamma B L_\pm$, with solution:

$$
L_\pm(t) = L_\pm(0) \, \mathrm{e}^{\mp \mathrm{i} \omega_L t}
$$

where $\omega_L = \gamma B$. The tip of $\boldsymbol{L}$ traces a circle in the $xy$-plane at frequency $\omega_L$, while $L_z$ remains constant.
:::

:::{admonition} Discussion
:class: dropdown tip

The precession equation $\mathrm{d}\boldsymbol{L}/\mathrm{d}t = \gamma \boldsymbol{L} \times \boldsymbol{B}$ has the same mathematical form as the Lorentz force equation $m\mathrm{d}\boldsymbol{v}/\mathrm{d}t = q\boldsymbol{v} \times \boldsymbol{B}$. What physical quantity plays the role of charge? What plays the role of velocity? Where does the analogy break down?
:::

### The Gyromagnetic Ratio

A spinning charged body generates a magnetic dipole moment. For a classical charge distribution with charge-to-mass ratio $q/m$ rotating rigidly:

$$
\boldsymbol{\mu} = \frac{q}{2m} \boldsymbol{L}
$$

The proportionality constant $\gamma = q/(2m)$ is the **gyromagnetic ratio**. We define the dimensionless **$g$-factor** by:

$$
\boldsymbol{\mu} = g \frac{q}{2m} \boldsymbol{S}
$$ (eq-g-factor)

:::{admonition} Classical vs Quantum $g$-factor
:class: important

| System | $g$-factor |
|--------|-----------|
| Classical orbital motion | $g = 1$ |
| Electron spin | $g \approx 2.002$ (Dirac equation gives exactly 2; QED corrections add 0.002) |
| Proton spin | $g \approx 5.586$ (anomalous, due to internal quark structure) |
| Neutron spin | $g \approx -3.826$ (nonzero despite zero charge — reveals internal structure) |

The electron's $g \approx 2$ cannot be explained classically and was one of the triumphs of the Dirac equation.
:::

:::{admonition} Derivation: Classical Gyromagnetic Ratio $g = 1$
:class: dropdown information

Consider a rigid body with uniform charge-to-mass ratio $q/m$ rotating with angular velocity $\boldsymbol{\omega}$. For a mass element $\mathrm{d}m$ at position $\boldsymbol{r}$ from the rotation axis:

**Magnetic moment contribution:** A charge element $\mathrm{d}q = (q/m)\mathrm{d}m$ moving in a circle of radius $r_\perp = |\boldsymbol{r} \times \hat{\boldsymbol{\omega}}|$ at speed $v = \omega r_\perp$ forms a current loop with magnetic moment:

$$
\mathrm{d}\mu = (\text{current}) \times (\text{area}) = \frac{\mathrm{d}q \cdot \omega}{2\pi} \cdot \pi r_\perp^2 = \frac{q}{2m} \omega r_\perp^2 \,\mathrm{d}m
$$

**Angular momentum contribution:** The same mass element contributes angular momentum $\mathrm{d}L = r_\perp^2 \omega \,\mathrm{d}m$.

Taking the ratio:

$$
\frac{\mathrm{d}\mu}{\mathrm{d}L} = \frac{q}{2m}
$$

Since this ratio is the *same* for every mass element (independent of $r_\perp$), it holds for the total quantities:

$$
\boldsymbol{\mu} = \frac{q}{2m}\boldsymbol{L} \implies g = 1
$$

This result holds for *any* rigid charge distribution with uniform $q/m$, regardless of shape. The electron's anomalous $g \approx 2$ is a purely quantum-relativistic effect (predicted by the Dirac equation).
:::

For an electron with charge $-e$ and spin $\boldsymbol{S}$:

$$
\boldsymbol{\mu}_e = -g_e \frac{e}{2m_e} \boldsymbol{S} = -g_e \mu_B \frac{\boldsymbol{S}}{\hbar}
$$

where $\mu_B = e\hbar/(2m_e)$ is the **Bohr magneton**.

### The Puzzle: Integer vs Half-Integer Angular Momentum

In quantum mechanics, orbital angular momentum $\hat{\boldsymbol{L}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{p}}$ acts on wavefunctions $\psi(\theta, \phi)$ on the unit sphere. The eigenvalue problem:

$$
\hat{L}^2 Y_l^m = \hbar^2 l(l+1) Y_l^m, \quad \hat{L}_z Y_l^m = \hbar m \, Y_l^m
$$

requires **single-valued wavefunctions** on the sphere: $\psi(\theta, \phi + 2\pi) = \psi(\theta, \phi)$. This forces $m \in \mathbb{Z}$, and consequently:

$$
l = 0, 1, 2, \ldots \quad \text{(integers only)}
$$

:::{admonition} Orbital Angular Momentum Is Always Integer
:class: important

Because wavefunctions must be single-valued on the sphere $S^2$, orbital angular momentum quantum numbers are restricted to non-negative integers. The eigenfunctions are the spherical harmonics $Y_l^m(\theta, \phi)$.
:::

:::{admonition} Derivation: Integer $l$ from Single-Valuedness
:class: dropdown information

The eigenvalue equation $\hat{L}_z \psi = \hbar m \,\psi$ in spherical coordinates reads:

$$
-\mathrm{i}\hbar \frac{\partial}{\partial \phi}\psi(\theta,\phi) = \hbar m\,\psi(\theta,\phi)
$$

The solution is $\psi(\theta,\phi) = \Theta(\theta)\,\mathrm{e}^{\mathrm{i} m\phi}$. Single-valuedness on the sphere requires $\psi(\theta,\phi + 2\pi) = \psi(\theta,\phi)$:

$$
\mathrm{e}^{\mathrm{i} m(\phi + 2\pi)} = \mathrm{e}^{\mathrm{i} m\phi} \implies \mathrm{e}^{2\pi\mathrm{i} m} = 1 \implies m \in \mathbb{Z}
$$

Since $m$ ranges from $-l$ to $+l$ in integer steps and must be an integer, $l$ must also be a non-negative integer: $l = 0, 1, 2, \ldots$

**Key point:** This argument uses only the topology of the configuration space ($S^2$ with its $2\pi$-periodic azimuthal coordinate). When a monopole gauge field is present (§4.4.2), the wavefunction acquires an additional $\phi$-dependent phase from the gauge connection, modifying the single-valuedness condition and allowing half-integer $l$.
:::

Yet the Stern-Gerlach experiment (1922) showed that silver atoms split into **two** beams in an inhomogeneous magnetic field — corresponding to $m = \pm 1/2$. This half-integer angular momentum cannot arise from any orbital motion of a particle on a sphere.

:::{admonition} The Stern-Gerlach Puzzle
:class: attention

The algebra $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar \epsilon_{ijk} \hat{J}_k$ allows both integer and half-integer representations ($j = 0, 1/2, 1, 3/2, \ldots$). But orbital angular momentum $\hat{\boldsymbol{L}} = \hat{\boldsymbol{r}} \times \hat{\boldsymbol{p}}$ only produces integer $l$, because wavefunctions on $S^2$ must be single-valued.

Half-integer spin was introduced as an additional postulate (Uhlenbeck and Goudsmit, 1925). But is it truly independent, or does it have a deeper origin?
:::

The resolution comes from **gauge topology**: when a particle moves on a sphere threaded by a magnetic monopole (§4.4.2), the single-valuedness condition is modified. The monopole's gauge field shifts the minimum angular momentum to $l_\text{min} = |s|$, where $s = qg/(2\hbar c)$ is the monopole quantum number. For $s = 1/2$, the lowest orbital states are a doublet with $l = 1/2$ — precisely spin-1/2. Thus half-integer angular momentum emerges from topology, not as an independent axiom.

### Summary

- A classical spinning top has orbital angular momentum $\boldsymbol{L} = \boldsymbol{r} \times \boldsymbol{p}$ and spin angular momentum $\boldsymbol{S}$ from body rotation
- A magnetic dipole in a field precesses at the Larmor frequency $\omega_L = \gamma B$, with the precession equation formally identical to the Lorentz force
- The gyromagnetic ratio $g$ relates magnetic moment to angular momentum; $g = 1$ classically, $g \approx 2$ for the electron (a relativistic quantum effect)
- Orbital angular momentum on a sphere is restricted to integers by single-valuedness of wavefunctions
- Half-integer angular momentum (spin-1/2) is observed experimentally but cannot arise from orbital motion — its origin lies in gauge topology (§4.4.2–4.4.3)
