## Lecture Notes

Gauge transformations represent one of the deepest principles in physics: the freedom to redefine quantum phases without changing physical predictions. This principle lies at the heart of gauge theory—the mathematical framework underlying all three fundamental forces (electromagnetic, weak, and strong). Understanding gauge transformations is the key to understanding why gauge fields exist and why they couple to matter in the way they do.

### Global Phase Symmetry

#### The Unobservable Global Phase

In quantum mechanics, the *absolute* phase of a wavefunction is never observable. Consider two wavefunctions that differ by a global phase:

$$


\psi'(x) = \mathrm{e}^{\mathrm{i}\alpha}\psi(x)

$$


where $\alpha$ is a constant (independent of position and time).

:::{admonition} Global Phase is Unobservable
:class: important

The Born rule for probability densities is invariant under global phase transformations:

$$


|\psi'(x)|^2 = |\mathrm{e}^{\mathrm{i}\alpha}\psi(x)|^2 = |\psi(x)|^2

$$


Therefore, global phase $\alpha$ has **no physical consequence**. Expectation values of all observables are identical:

$$


\langle\psi'|\hat{O}|\psi'\rangle = \langle\psi|\hat{O}|\psi\rangle

$$


This is the first gauge principle: **global phase redundancy is fundamental to quantum mechanics**.
:::

#### Physical Interpretation

Why is global phase unobservable? Because:
1. **Energy is only defined up to an arbitrary zero point.** Shifting $\hat{H} \to \hat{H} + E_0$ changes all energies but physics remains identical.
2. **Global phase is conjugate to total particle number.** Only *differences* in phase matter; absolute values are meaningless.
3. **Quantum mechanics is about correlations, not absolute states.** The global phase of a single system is inaccessible; only relative phases between components of a superposition matter.

### Local Gauge Transformations: The Problem

Now consider a *position-dependent* phase transformation:

$$


\psi(x,t) \to \psi'(x,t) = \mathrm{e}^{\mathrm{i}\chi(x,t)}\psi(x,t)

$$


where $\chi(x,t)$ depends on space and time. This is called a **local phase transformation** or **local U(1) gauge transformation**.

#### Born Rule is Preserved

The Born rule is still invariant:

$$


|\psi'(x,t)|^2 = |\mathrm{e}^{\mathrm{i}\chi(x,t)}\psi(x,t)|^2 = |\psi(x,t)|^2

$$


So local phase changes, like global phase changes, do not affect probability densities.

#### The Kinetic Energy Problem

But the kinetic energy term is *not* invariant. The spatial derivative of $\psi'$ is:

$$


\partial_i \psi' = \partial_i(\mathrm{e}^{\mathrm{i}\chi}\psi) = \mathrm{e}^{\mathrm{i}\chi}(\partial_i\chi)\psi + \mathrm{e}^{\mathrm{i}\chi}(\partial_i\psi) = \mathrm{e}^{\mathrm{i}\chi}(\partial_i\chi + \partial_i)\psi

$$


The kinetic energy in the Schrödinger equation:

$$


\hat{T} = -\frac{\hbar^2}{2m}\nabla^2

$$


depends on second derivatives, and transforms as:

$$


\hat{T}\psi' \neq \mathrm{e}^{\mathrm{i}\chi}\hat{T}\psi

$$


:::{admonition} The Gauge Problem
:class: important

Under a local gauge transformation $\psi \to \mathrm{e}^{\mathrm{i}\chi}\psi$, the **Schrödinger equation is not covariant**:

$$


\mathrm{i}\hbar\frac{\partial\psi'}{\partial t} \neq -\frac{\hbar^2}{2m}\nabla^2\psi' + V(\boldsymbol{r})\psi'

$$


The kinetic energy term gains extra contributions involving derivatives of $\chi$, destroying the form of the equation.

This is the **gauge problem**: if local phase is truly unobservable (like global phase), why does changing it break the fundamental equation?
:::

### Gauge Fields and Covariant Derivatives

The resolution is profound: **we must introduce a new field—the gauge field—to restore covariance**.

#### The Covariant Derivative

Define the **covariant derivative**:

$$


D_i = \partial_i + \mathrm{i}qA_i

$$


where $q$ is the charge of the particle and $A_i(\boldsymbol{r},t)$ is a new field (the **gauge field**).

:::{admonition} Gauge Field Transformation
:class: important

Under a local gauge transformation $\psi \to \mathrm{e}^{\mathrm{i}q\chi}\psi$, the gauge field must transform as:

$$


A_i \to A_i - \partial_i\chi

$$


(Note: The factor of $q$ in the exponent ensures dimensional consistency with $A_i$.)

With this transformation, the covariant derivative transforms covariantly:

$$


D_i\psi \to \mathrm{e}^{\mathrm{i}q\chi}D_i\psi

$$


:::

#:::{admonition} Derivation: Covariance of the Covariant Derivative
:class: dropdown

Under $\psi \to \psi' = \mathrm{e}^{\mathrm{i}q\chi}\psi$ and $A_i \to A_i' = A_i - \partial_i\chi$:

$$


D_i'\psi' = (\partial_i + \mathrm{i}qA_i')\psi' = (\partial_i + \mathrm{i}q(A_i - \partial_i\chi))(\mathrm{e}^{\mathrm{i}q\chi}\psi)

$$


$$

= \partial_i(\mathrm{e}^{\mathrm{i}q\chi}\psi) + \mathrm{i}qA_i(\mathrm{e}^{\mathrm{i}q\chi}\psi) - \mathrm{i}q(\partial_i\chi)(\mathrm{e}^{\mathrm{i}q\chi}\psi)

$$


$$

= (\mathrm{i}q\partial_i\chi)\mathrm{e}^{\mathrm{i}q\chi}\psi + \mathrm{e}^{\mathrm{i}q\chi}(\partial_i\psi) + \mathrm{i}qA_i(\mathrm{e}^{\mathrm{i}q\chi}\psi) - \mathrm{i}q(\partial_i\chi)(\mathrm{e}^{\mathrm{i}q\chi}\psi)

$$


The first and last terms cancel:

$$


D_i'\psi' = \mathrm{e}^{\mathrm{i}q\chi}[(\partial_i + \mathrm{i}qA_i)\psi] = \mathrm{e}^{\mathrm{i}q\chi}D_i\psi

$$


This is the fundamental property: **the covariant derivative transforms exactly like $\psi$ itself**.
:::

### Gauge-Covariant Schrödinger Equation

Replace all ordinary derivatives with covariant derivatives in the Schrödinger equation. The **minimal coupling** prescription states:

$$


\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}

$$


This gives the **gauge-covariant Schrödinger equation**:

$$


\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + V + q\Phi\right]\psi

$$


where $\Phi(\boldsymbol{r},t)$ is the scalar potential (time component of the gauge field), and we've added the Coulomb potential $q\Phi$ that couples the charge to the scalar field.

:::{admonition} Note: Why Minimal Coupling?
:class: note

The minimal coupling prescription is the **simplest** (lowest-order) way to couple charges to gauge fields. Higher-order couplings $\propto \nabla^2\boldsymbol{A}$ are possible but are generally suppressed for slowly-varying fields. The name "minimal" refers to this simplicity.

Quantum field theory interprets minimal coupling as arising from **gauge invariance**—it is the unique way to maintain local gauge symmetry. This transforms gauge symmetry from a mathematical redundancy into a **principle that determines physics**.
:::

:::{admonition} Gauge Equivalence
:class: important

Two pairs $(\psi, \boldsymbol{A})$ and $(\psi', \boldsymbol{A}')$ related by a gauge transformation are **physically equivalent**: they describe the same quantum state and produce identical predictions for all observables.

The set of all gauge transformations forms a group—for the case of local U(1) phase rotations, this is the **U(1) gauge group**. Each equivalence class under gauge transformations corresponds to one physical state.
:::

#### Degrees of Freedom

In 3D space, the gauge field $\boldsymbol{A}(\boldsymbol{r})$ has 3 components (it's a vector). But only **2 physical degrees of freedom** correspond to transverse electromagnetic waves (the two polarizations). The "extra" degree of freedom is removed by gauge freedom—we can always choose a gauge condition (like the Coulomb gauge $\nabla \cdot \boldsymbol{A} = 0$) to eliminate the unphysical longitudinal component.

:::{admonition} Example: Coulomb Gauge
:class: example

In the **Coulomb gauge**, we impose the constraint:

$$


\nabla \cdot \boldsymbol{A} = 0

$$


This is a *gauge choice*—we use our gauge freedom to eliminate the divergence of $\boldsymbol{A}$. Different gauge choices lead to different forms of the equations, but all physical predictions remain identical.

In the Coulomb gauge, the scalar potential $\Phi$ is determined instantaneously by the charge density (no dynamic degrees of freedom), while $\boldsymbol{A}$ satisfies a wave equation with sources from the current.

This gauge is natural for atomic systems where we want to emphasize the instantaneous Coulomb interaction between charges.
:::

### Discussion: Global Phase vs. Local Gauge

The discussion box in this section addresses the profound question:

:::{admonition} Discussion: Global Phase vs Local Gauge
:class: dropdown warning

A global phase change $\psi \to \mathrm{e}^{\mathrm{i}\alpha}\psi$ is unobservable. But a position-dependent phase $\mathrm{e}^{\mathrm{i}\alpha(x)}\psi$ requires introducing a gauge field. 

**Calculate** the force law that emerges from requiring the Schrödinger equation to remain covariant under local phase transformations, and show it equals the Lorentz force.

This reveals why gauge symmetry is not just mathematical—it is the principle that *forces* gauge fields (and hence electromagnetic interactions) to exist.
:::

This is one of the deepest insights of modern physics: **gauge principles are not imposed as extra structure—they flow naturally from the requirement that physics be independent of unobservable phase choices**.

### Summary

- A **global phase** change $\psi \to \mathrm{e}^{\mathrm{i}\alpha}\psi$ is completely unobservable—probability and all expectation values are unchanged.
- A **local phase** transformation $\psi \to \mathrm{e}^{\mathrm{i}q\chi(x,t)}\psi$ is a priori unobservable (it shouldn't change physics). But without gauge fields, it violates the Schrödinger equation.
- To maintain covariance, we must introduce a **gauge field** $\boldsymbol{A}$ that transforms as $\boldsymbol{A} \to \boldsymbol{A} - \nabla\chi$.
- The **covariant derivative** $D_i = \partial_i + \mathrm{i}qA_i$ transforms covariantly: $D_i\psi \to \mathrm{e}^{\mathrm{i}q\chi}D_i\psi$.
- The **gauge-covariant Schrödinger equation** is obtained by the minimal coupling prescription: $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$.
- Gauge transformations are a **redundancy**: different gauge choices describe the same physical state.
- Gauge symmetry is a fundamental principle that *determines* how charges couple to electromagnetic fields.
