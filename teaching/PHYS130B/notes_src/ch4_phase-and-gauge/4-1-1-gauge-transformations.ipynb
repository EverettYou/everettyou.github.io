## Lecture Notes

The global phase of a quantum wavefunction is unobservable—a fundamental symmetry of quantum mechanics. Yet this innocent observation leads to a profound principle: local phase rotations must also leave physics unchanged, and this demand forces the existence of gauge fields that couple to all charged particles. In this section, we begin our exploration of gauge symmetry by understanding why the global phase is meaningless and why insisting on local gauge invariance is not merely a mathematical choice, but the deepest organizing principle underlying all fundamental forces.

### Observability and the Phase Ambiguity

The fundamental postulate of quantum mechanics relates physical probabilities to the wavefunction through the Born rule: $P(x, t) = |\psi(x, t)|^2$. This expression depends only on the **magnitude** of $\psi$, not its phase. When we rotate the wavefunction's global phase by a constant amount $\alpha$, giving $\psi' = e^{i\alpha}\psi$, all probabilities remain exactly unchanged: $P'(x, t) = |e^{i\alpha}|^2|\psi|^2 = |\psi|^2 = P(x, t)$. Similarly, expectation values of all observables remain unchanged. This is the **global phase ambiguity** or **global $U(1)$ gauge symmetry**, where $U(1)$ refers to the unitary group of phase rotations—a one-parameter family of transformations forming a circle in the complex plane.

:::{admonition} Derivation: Phase Invariance of the Born Rule
:class: dropdown note

The fundamental postulate of quantum mechanics relates physical probabilities to the wavefunction through:

$$

P(x, t) = |\psi(x, t)|^2 = \psi^*(x, t)\psi(x, t)

$$

This expression depends only on the squared **magnitude** of $\psi$, not its phase. Consider what happens when we rotate the wavefunction's phase by a constant amount $\alpha$:

$$

\psi(x, t) \to \psi'(x, t) = e^{i\alpha}\psi(x, t)

$$

where $\alpha$ is a real number independent of position and time. The probability becomes:

$$

P'(x, t) = |\psi'(x, t)|^2 = |e^{i\alpha}|^2 |\psi(x, t)|^2 = |\psi(x, t)|^2 = P(x, t)

$$

**All probabilities remain unchanged.** Expectation values of observables also remain unchanged:

$$

\langle A \rangle' = \int (\psi')^* \hat{A} \psi'\, dx = e^{-i\alpha}e^{i\alpha}\int \psi^* \hat{A} \psi\, dx = \langle A \rangle

$$

**Conclusion**: The wavefunction $\psi(x, t)$ and $e^{i\alpha}\psi(x, t)$ represent the **exact same physical state** and must be considered equivalent. There is no experiment that can distinguish between them. This is a fundamental symmetry of quantum mechanics.

:::

:::{admonition} Global vs Local
:class: note

The term "global" means the phase rotation $e^{i\alpha}$ is the **same everywhere**—the same angle $\alpha$ applies at all spatial points. This contrasts sharply with local transformations, where the angle varies with position.

:::

### The Failure of Local Gauge Invariance

Now consider a **position-dependent** phase rotation: $\psi(x, t) \to \psi'(x, t) = e^{i\chi(x,t)}\psi(x, t)$, where $\chi(x, t)$ is an arbitrary real function. The probability distribution still transforms as $P'(x, t) = |e^{i\chi(x,t)}|^2|\psi(x, t)|^2 = |\psi(x, t)|^2$—the local transformation **preserves all probability densities!** This seems to suggest that local gauge freedom should also be unobservable.

However, **this naive expectation is wrong.** The problem lies in spatial derivatives. The standard Schrödinger equation contains kinetic energy terms with spatial derivatives:

$$

i\hbar\frac{\partial\psi}{\partial t} = \left[\frac{-\hbar^2}{2m}\nabla^2 + V(x, t)\right]\psi

$$

When we apply a local gauge transformation, the kinetic term transforms in a problematic way that **breaks the equation of motion.**

:::{admonition} Calculation: Spatial Derivatives Under Local Gauge Transformation
:class: dropdown note

Let's compute how the spatial derivative of the wavefunction transforms:

$$

\partial_i\psi'(x, t) = \partial_i[e^{i\chi(x,t)}\psi(x, t)]

$$

Using the product rule:

$$

\partial_i\psi' = (\partial_i e^{i\chi})\psi + e^{i\chi}(\partial_i\psi)

$$

$$

\partial_i\psi' = e^{i\chi}(i\partial_i\chi)\psi + e^{i\chi}(\partial_i\psi)

$$

$$

\partial_i\psi' = e^{i\chi}[i(\partial_i\chi)\psi + \partial_i\psi]

$$

The key observation: **an extra term $(i\partial_i\chi)$ appears!** When we apply the kinetic energy operator:

$$

\nabla^2\psi' = \nabla \cdot (e^{i\chi}[(i\nabla\chi)\psi + \nabla\psi])

$$

The Schrödinger equation transforms as:

$$

i\hbar\frac{\partial\psi'}{\partial t} = e^{i\chi}\left[i\hbar\frac{\partial\chi}{\partial t}\psi + i\hbar\frac{\partial\psi}{\partial t}\right]

$$

$$

\frac{-\hbar^2}{2m}\nabla^2\psi' = e^{i\chi}\left[\frac{-\hbar^2}{2m}(\nabla^2\psi) + \text{(unwanted terms involving } \nabla\chi)\right]

$$

**The equation of motion is destroyed by the appearance of gradient terms from the phase function!** The standard Schrödinger equation is **NOT gauge covariant**—it doesn't maintain its form under local gauge transformations.

:::

### The Covariant Derivative: Restoring Gauge Covariance

To restore gauge covariance, we must replace ordinary derivatives with **covariant derivatives** that incorporate a gauge field. Define the covariant derivative as:

$$

D_i = \partial_i + iqA_i(x, t)

$$

where $A_i(x, t)$ is a new field we introduce—the **gauge field** or **gauge connection**. The dimensionless coupling $q$ is the charge parameter (for electromagnetism, $q = e$ for an electron). The power of this construction is that $D_i$ **transforms like $\psi$ itself** under gauge transformations: if the gauge field transforms as $A_i(x, t) \to A_i'(x, t) = A_i(x, t) - \frac{1}{q}\partial_i\chi(x, t)$, then the covariant derivative transforms exactly like the wavefunction. This is the definition of **gauge covariance**.

:::{admonition} Proof: Gauge Covariance of the Covariant Derivative
:class: dropdown note

If the gauge field transforms as:

$$

A_i(x, t) \to A_i'(x, t) = A_i(x, t) - \frac{1}{q}\partial_i\chi(x, t)

$$

then:

$$

D_i\psi \to D_i'\psi' = (\partial_i + iqA_i')e^{i\chi}\psi

$$

$$

= e^{i\chi}\partial_i\psi + i(\partial_i\chi)e^{i\chi}\psi + iqe^{i\chi}\left(A_i - \frac{1}{q}\partial_i\chi\right)\psi

$$

$$

= e^{i\chi}\partial_i\psi + i(\partial_i\chi)e^{i\chi}\psi + iqe^{i\chi}A_i\psi - i(\partial_i\chi)e^{i\chi}\psi

$$

$$

= e^{i\chi}(\partial_i + iqA_i)\psi = e^{i\chi}D_i\psi

$$

**Perfect!** The covariant derivative transforms exactly like the wavefunction:

$$

D_i\psi \to e^{i\chi}D_i\psi

$$

:::

### Gauge-Covariant Schrödinger Equation

The gauge-covariant form of the Schrödinger equation replaces the ordinary kinetic energy operator with the covariant derivative:

$$

i\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(\boldsymbol{D} - i\partial_t)^2 + V\right]\psi

$$

or in the more standard form with scalar potential $\Phi(x, t)$:

$$

i\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(-i\hbar\boldsymbol{D})^2 + q\Phi(x, t)\right]\psi

$$

The canonical momentum operator transforms to the **kinetic momentum**:

$$

\hat{\boldsymbol{p}} = -i\hbar\nabla \to \hat{\boldsymbol{\pi}} = -i\hbar\boldsymbol{D} = -i\hbar\nabla - q\boldsymbol{A}

$$

This gives the **minimal coupling** prescription: $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$. This simple rule tells us how to modify any quantum system in the presence of external gauge fields—a fundamental principle that applies across all of physics.

:::{admonition} Details: Minimal Coupling Prescription
:class: dropdown note

The gauge-covariant form of the Schrödinger equation is:

$$

i\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(\boldsymbol{D} - i\partial_t)^2 + V\right]\psi

$$

or more commonly written in terms of the full covariant 4-derivative:

$$

i\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(-i\hbar\boldsymbol{D})^2 + q\Phi(x, t)\right]\psi

$$

where $\Phi(x, t)$ is the scalar potential (for electromagnetism, the electric potential).

The canonical momentum operator is:

$$

\hat{\boldsymbol{p}} = -i\hbar\nabla \to \hat{\boldsymbol{\pi}} = -i\hbar\boldsymbol{D} = -i\hbar\nabla - q\boldsymbol{A}

$$

This gives the **minimal coupling** prescription:

$$

\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}

$$

:::

### Gauge Freedom and Physical Equivalence

The remarkable feature of gauge theories is that **multiple gauge field configurations describe the same physics**. If two gauge fields $\boldsymbol{A}$ and $\boldsymbol{A}'$ are related by:

$$

\boldsymbol{A}'(x, t) = \boldsymbol{A}(x, t) - \frac{1}{q}\nabla\chi(x, t)

$$

and correspondingly $\Phi' = \Phi + \frac{1}{q}\partial_t\chi$, then the physics is unchanged. We can choose the "easiest" gauge field to work with in any given problem—a tremendous practical advantage.

### Observable vs Gauge-Dependent Quantities

In gauge theory, we must distinguish carefully between:

**Gauge-dependent quantities** (unobservable):
- The wavefunction phase $\chi(x, t)$
- The vector potential $\boldsymbol{A}(x, t)$
- The scalar potential $\Phi(x, t)$

**Gauge-invariant quantities** (observable):
- Probability density $|\psi|^2$
- Expectation values of physical operators
- Line integrals of the gauge field around closed loops
- The field strength tensor (to be defined in the next section)

The gauge field potentials are **fundamental** to quantum mechanics—they directly affect the quantum wavefunction evolution. Yet, any specific choice of potential is unobservable; only gauge-invariant combinations matter.

:::{admonition} Key Insight
:class: important

Gauge theory is not just a mathematical convenience—it is a fundamental principle. Local gauge symmetry **requires** the existence of gauge fields, which in turn couple to charged particles and determine their quantum dynamics. This deep principle explains the origin of fundamental forces.

:::

### Semiclassical Equations of Motion

In the semiclassical approximation, the wavefunction's phase contains the classical action: $\psi(x, t) = A(x, t) e^{iS(x, t)/\hbar}$, where $S$ is the classical action and $A$ is the amplitude. For the gauge-covariant Schrödinger equation, the leading order equations in the WKB expansion are the **Hamilton-Jacobi equation** (from the phase):

$$

\frac{\partial S}{\partial t} + \frac{1}{2m}(\nabla S - q\boldsymbol{A})^2 + q\Phi = 0

$$

This is the classical energy conservation equation. Defining the velocity as $\boldsymbol{v} = \frac{1}{m}(\nabla S - q\boldsymbol{A})$, we recover the **Lorentz force equation**:

$$

m\frac{d\boldsymbol{v}}{dt} = -\nabla(q\Phi) + q\boldsymbol{v} \times (\nabla \times \boldsymbol{A}) = q(\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B})

$$

This shows that the gauge-covariant formulation guarantees the correct classical limit, including electromagnetic forces. The gauge field connection enters through both the kinetic momentum correction and through the field strength tensor $\boldsymbol{B} = \nabla \times \boldsymbol{A}$.

:::{admonition} Derivation: WKB Approximation and Classical Limit
:class: dropdown note

The semiclassical approximation describes the classical limit of quantum mechanics, where the wavefunction's phase contains the action. We write:

$$

\psi(x, t) = A(x, t) e^{i S(x, t)/\hbar}

$$

where $S$ is the classical action and $A$ is the amplitude. For the gauge-covariant Schrödinger equation:

$$

i\hbar\frac{\partial\psi}{\partial t} = \left[\frac{1}{2m}(-i\hbar\nabla - q\boldsymbol{A})^2 + q\Phi\right]\psi

$$

Substituting the WKB form and expanding in powers of $\hbar$, the leading order equations are:

**Hamilton-Jacobi equation (from the phase):**

$$

\frac{\partial S}{\partial t} + \frac{1}{2m}(\nabla S - q\boldsymbol{A})^2 + q\Phi = 0

$$

This is the classical energy conservation equation. Define the velocity as $\boldsymbol{v} = \frac{1}{m}(\nabla S - q\boldsymbol{A})$. Then:

**Equation of motion (Newton's second law):**

$$

m\frac{d\boldsymbol{v}}{dt} = -\nabla(q\Phi) + q\boldsymbol{v} \times (\nabla \times \boldsymbol{A}) = q(\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B})

$$

This is exactly the **Lorentz force equation**. The gauge-covariant formulation guarantees that the classical limit recovers the correct equations of motion, including electromagnetic forces. The gauge field connection $\boldsymbol{A}$ enters through both the canonical momentum correction and through the field strength tensor $\boldsymbol{B} = \nabla \times \boldsymbol{A}$.

:::

### Geometric Phase and the Berry Phase

When a quantum system evolves adiabatically through a closed loop in parameter space, its wavefunction acquires an additional phase beyond the dynamical phase from the Hamiltonian. This is the **Berry phase** or **geometric phase**, and it provides a direct physical manifestation of gauge structure in quantum mechanics. Consider a system with instantaneous eigenstates $|n(\boldsymbol{R}(t))\rangle$ where $\boldsymbol{R}(t)$ are slowly varying parameters. The Berry phase accumulated along a closed path $\mathcal{C}$ is:

$$

\gamma_n = i \oint_\mathcal{C} \langle n(\boldsymbol{R})|\nabla_{\boldsymbol{R}} n(\boldsymbol{R})\rangle \cdot d\boldsymbol{R}

$$

This has the same form as a line integral of a gauge field in parameter space! The **Berry connection** $\mathcal{A}_\boldsymbol{R} = i\langle n|\nabla_{\boldsymbol{R}} n\rangle$ plays exactly the role of a gauge field—the phase acquired depends on the path through parameter space and exhibits the same topological properties as a gauge theory.

:::{admonition} Derivation: Berry Phase from the Adiabatic Theorem
:class: dropdown note

Start with the Schrödinger equation:

$$

i\hbar\frac{\partial|\psi\rangle}{\partial t} = H(\boldsymbol{R}(t))|\psi\rangle

$$

Under adiabatic evolution, the state remains in the $n$-th eigenstate:

$$

|\psi(t)\rangle = e^{i\phi_n(t)}e^{i\gamma_n(t)}|n(\boldsymbol{R}(t))\rangle

$$

where $\phi_n(t) = -\frac{1}{\hbar}\int_0^t E_n(\boldsymbol{R}(t'))dt'$ is the **dynamical phase** and $\gamma_n$ is the geometric phase.

Substituting into the Schrödinger equation and using $H(\boldsymbol{R})|n(\boldsymbol{R})\rangle = E_n(\boldsymbol{R})|n(\boldsymbol{R})\rangle$:

$$

i\hbar\left[i\dot{\phi}_n + i\dot{\gamma}_n + |n\rangle\langle n|\partial_t + (\partial_t |n\rangle)\langle n|\right]e^{i\phi_n}e^{i\gamma_n}|n\rangle = H|\psi\rangle

$$

The dynamical phase term $\dot{\phi}_n = -E_n(\boldsymbol{R})/\hbar$ cancels with the eigenvalue. The remaining terms give:

$$

i\hbar\dot{\gamma}_n = i\hbar\langle n|\partial_t n\rangle = i\hbar\langle n|\nabla_\boldsymbol{R} n\rangle \cdot \dot{\boldsymbol{R}}

$$

Therefore:

$$

\dot{\gamma}_n = \langle n(\boldsymbol{R})|i\nabla_{\boldsymbol{R}} n(\boldsymbol{R})\rangle \cdot \dot{\boldsymbol{R}}

$$

Integrating around a closed path:

$$

\gamma_n = i \oint_\mathcal{C} \langle n|\nabla_{\boldsymbol{R}} n\rangle \cdot d\boldsymbol{R}

$$

**Physical interpretation:** The Berry connection $\mathcal{A}_\boldsymbol{R} = i\langle n|\nabla_{\boldsymbol{R}} n\rangle$ plays the same role as a gauge field in parameter space. The Berry phase is a line integral of this connection, analogous to the Aharonov-Bohm phase acquired when a charged particle moves through an electromagnetic potential.

:::

### Summary

- **Phase ambiguity**: Wavefunctions differing by global phase $e^{i\alpha}$ are physically equivalent; only probability densities are observable.
- **Global $U(1)$ symmetry**: The simplest symmetry—phase rotations by a constant angle—is a fundamental feature of quantum mechanics.
- **Local gauge transformation**: Position-dependent phase rotation $\psi \to e^{i\chi(x,t)}\psi$ preserves probabilities but breaks the Schrödinger equation's form.
- **Gauge field necessity**: To maintain covariance under local transformations, new fields $A_i(x, t)$ must be introduced.
- **Covariant derivative**: $D_i = \partial_i + iqA_i$ transforms like the wavefunction and ensures gauge covariance.
- **Minimal coupling**: The rule $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$ encodes the interaction between charged particles and gauge fields.
- **Gauge freedom**: Multiple field configurations (differing by gauge transformations) represent identical physics; we can choose the most convenient gauge.
- **Observables are gauge-invariant**: Only quantities independent of gauge choice have physical meaning.
