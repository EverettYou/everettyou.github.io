## Lecture Notes

Gauge freedom is not a bug in gauge theory—it is a feature. By choosing gauges cleverly, we can dramatically simplify calculations while remaining confident that physical observables are gauge-invariant. This section showcases practical strategies: the Coulomb gauge decouples radiation from Coulomb interactions, the temporal gauge clarifies physical vs unphysical degrees of freedom, and other gauges reveal hidden symmetries. Mastery of gauge freedom is essential for solving real problems in quantum mechanics and field theory.
### Gauge Freedom and Computational Strategy

Gauge freedom is a fundamental feature of electromagnetism: the vector potential $\boldsymbol{A}$ and scalar potential $\Phi$ are not unique. Infinitely many choices of potentials describe the same physical fields:

$$

\boldsymbol{B} = \nabla \times \boldsymbol{A}, \quad \boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}

$$

Rather than a limitation, this is a powerful computational tool—**we choose the gauge that makes calculations easiest**. Different gauges reveal different aspects of the physics. By exploiting gauge freedom, we can dramatically simplify problems while maintaining confidence that physical predictions are identical regardless of gauge choice.

:::{admonition} Derivation: Gauge Transformations
:class: dropdown note


A gauge transformation changes potentials while leaving fields unchanged:

$$

\boldsymbol{A} \to \boldsymbol{A}' = \boldsymbol{A} + \nabla\chi, \quad \Phi \to \Phi' = \Phi - \partial_t\chi

$$

where $\chi(\boldsymbol{r}, t)$ is an arbitrary scalar function. The fields are unchanged:

$$

\boldsymbol{B}' = \nabla \times \boldsymbol{A}' = \nabla \times (\boldsymbol{A} + \nabla\chi) = \nabla \times \boldsymbol{A} = \boldsymbol{B}

$$

The wavefunction also transforms: $\psi \to e^{iq\chi/\hbar c}\psi$, ensuring the Schrödinger equation and all physical predictions remain invariant.

:::

### Common Gauge Choices

**Coulomb Gauge**: $\nabla \cdot \boldsymbol{A} = 0$
- Simplifies the Schrödinger equation (no divergence terms)
- Scalar potential $\Phi$ satisfies Poisson's equation (instantaneous Coulomb potential)
- Useful for non-relativistic problems with time-varying fields

**Landau Gauge**: $\boldsymbol{A} = B_0 y \hat{x}$ (for uniform field $\boldsymbol{B} = B_0\hat{z}$)
- Breaks translational symmetry but simplifies 2D Landau level problems
- Hamiltonian separates into independent oscillator modes

**Symmetric Gauge**: $\boldsymbol{A} = \frac{1}{2}\boldsymbol{B} \times \boldsymbol{r}$
- Maintains rotational symmetry (useful for cylindrical/spherical problems)
- Natural for high-symmetry Landau level calculations

**Lorenz Gauge**: $\nabla \cdot \boldsymbol{A} + \frac{1}{c^2}\partial_t\Phi = 0$
- Manifestly Lorentz-covariant for relativistic problems
- Decouples wave equations for $\boldsymbol{A}$ and $\Phi$

### Wilson Loops: Gauge-Invariant Observables

Although the vector potential $\boldsymbol{A}$ is gauge-dependent, certain **non-local** observables built from $\boldsymbol{A}$ are gauge-invariant. The most important is the **Wilson loop**, which encodes the phase acquired by a quantum particle transported around a closed path:

$$

W(C) = \exp\left[i\frac{q}{\hbar c}\oint_C \boldsymbol{A} \cdot d\boldsymbol{l}\right]

$$

This is a **gauge-invariant observable**. Under a gauge transformation $\boldsymbol{A} \to \boldsymbol{A} - \nabla\chi$, the contour integral of the gradient term vanishes ($\oint \nabla\chi \cdot d\boldsymbol{l} = 0$ for single-valued $\chi$), leaving $W(C)$ unchanged. For a closed loop enclosing magnetic flux $\Phi_B$:

$$

W(C) = e^{i\frac{q\Phi_B}{\hbar c}}

$$

The **phase (argument) of the Wilson loop** encodes the enclosed flux in a completely gauge-invariant way. This is a beautiful demonstration that while individual potentials are gauge-dependent, their circulations encode observable physics.

:::{admonition} Derivation: Gauge Invariance of Wilson Loops
:class: dropdown note


Under gauge transformation $\boldsymbol{A} \to \boldsymbol{A} - \nabla\chi$:

$$

\oint_C (\boldsymbol{A} - \nabla\chi) \cdot d\boldsymbol{l} = \oint_C \boldsymbol{A} \cdot d\boldsymbol{l} - \oint_C \nabla\chi \cdot d\boldsymbol{l}

$$

The second term vanishes because $\chi$ is single-valued. Therefore:

$$

W(C) \text{ is gauge-invariant.}

$$

In non-abelian gauge theories (Yang-Mills, QCD), Wilson loops become matrix-valued and detect confinement, showing that gauge structures encode fundamental physics.

:::

### Peierls Substitution: Magnetic Fields in Lattice Models

When electrons hop on a crystal lattice in the presence of a magnetic field, each hopping term acquires a **phase factor** encoding the vector potential. This is the **Peierls substitution**—a microscopic mechanism for how magnetic fields affect electron dynamics on lattices.

In a tight-binding model without the field, nearest-neighbor hopping is:

$$

\hat{H}_0 = -t\sum_{\langle i,j\rangle}(c_i^\dagger c_j + \text{h.c.})

$$

When a magnetic field is applied, each hopping acquires a Peierls phase:

$$

c_i^\dagger c_j \to e^{i\phi_{ij}} c_i^\dagger c_j, \quad \phi_{ij} = \frac{q}{\hbar c}\int_i^j \boldsymbol{A} \cdot d\boldsymbol{l}

$$

This substitution automatically incorporates magnetic field effects, gauge invariance, and correct Lorentz force dynamics. In this framework, Landau-level-like behavior emerges from the lattice geometry—this is the origin of the quantum Hall effect in tight-binding models.

:::{admonition} Derivation: Peierls Phases in a Uniform Field
:class: dropdown note


For a uniform field $\boldsymbol{B} = B\hat{z}$ on a 2D square lattice with lattice constant $a$, using Landau gauge $\boldsymbol{A} = (0, Bx, 0)$:

**Horizontal hopping** from $(m,n)$ to $(m+1,n)$:

$$

\phi_{\to} = \frac{qBa}{\hbar c} \cdot n

$$

**Vertical hopping** from $(m,n)$ to $(m,n+1)$:

$$

\phi_{\uparrow} = -\frac{qBa}{\hbar c}\cdot (m+1/2)

$$

These phases depend on position—they encode the spatial variation of the vector potential. The observable physics (spectrum, transport) is gauge-invariant, even though individual phases change under gauge transformation.

:::

### Gauge Invariance of Physical Observables

A fundamental theorem of gauge theory states: **All physically measurable quantities are gauge-invariant.** This means that while we can choose any gauge to simplify calculations, the final physical predictions (probabilities, energies, currents) are independent of that choice.

For any observable operator that doesn't explicitly involve the gauge field:

$$

\langle\psi'|\hat{O}|\psi'\rangle = \langle\psi|\hat{O}|\psi\rangle

$$

even after gauge transformations. This protects all physical predictions: probability densities, energy eigenvalues, conductances, magnetic moments, and any other measured quantity remain gauge-invariant.

:::{admonition} Derivation: Gauge Invariance of Expectation Values
:class: dropdown note


Under a gauge transformation, the wavefunction transforms as:

$$

\psi \to \psi' = e^{iq\chi/\hbar c}\psi

$$

The expectation value of an observable $\hat{O}$ becomes:

$$

\langle\psi'|\hat{O}|\psi'\rangle = \int (e^{iq\chi/\hbar c}\psi)^*\hat{O}(e^{iq\chi/\hbar c}\psi)d\boldsymbol{r}

$$

$$

= \int e^{-iq\chi/\hbar c}\psi^*\hat{O}e^{iq\chi/\hbar c}\psi\, d\boldsymbol{r} = \int\psi^*\hat{O}\psi\, d\boldsymbol{r}

$$

The phase factors cancel (since the exponent is real and acts multiplicatively), proving gauge invariance.

:::{admonition} Gauge-Invariant vs Gauge-Dependent Observables
:class: example

**Gauge-invariant** (observable):
- Probability density: $|\psi|^2$
- Energy eigenvalues: $E_n$
- Conductance: $\sigma = dI/dV$
- Magnetic moments: $\langle\mu\rangle$
- Phase around closed loops via Wilson loops

**Gauge-dependent** (unobservable):
- Vector potential: $\boldsymbol{A}(\boldsymbol{r})$
- Scalar potential: $\Phi(\boldsymbol{r})$
- Wavefunction phase: $\arg[\psi(\boldsymbol{r})]$

:::

### Experimental Determination of Gauge-Invariant Quantities

**Interference patterns**: Directly reveal phase differences between paths—these encode enclosed flux (Aharonov-Bohm effect).

**Hall conductance**: Quantized at $\sigma_H = \nu e^2/h$, depending only on filling factor $\nu$ and fundamental constants, not on material details.

**Flux quantization**: Superconducting rings exhibit discrete flux values (integer multiples of $\Phi_0$), directly measurable via current response.

**SQUID measurements**: The oscillation period of critical current vs. flux directly reveals the flux quantum and tests gauge-invariant predictions.

**Spectroscopy**: Energy level splittings and transition frequencies depend only on gauge-invariant combinations of fields, independent of potential choice.

### Summary

- **Gauge freedom**: Choice of potentials $\boldsymbol{A}$ and $\Phi$ is arbitrary; different choices represent the same physics.
- **Gauge choices**: Coulomb, Landau, symmetric, and Lorenz gauges each simplify specific problem types.
- **Wilson loops**: Non-local observables $\exp[iq\oint\boldsymbol{A}\cdot d\boldsymbol{l}]$ that are completely gauge-invariant and encode enclosed flux.
- **Peierls substitution**: Method to incorporate magnetic fields into lattice models via phase factors on hopping terms.
- **Observable gauge-invariance**: All physical predictions (energies, forces, conductances, probabilities) are independent of gauge choice.
- **Unobservable potentials**: The potentials $\boldsymbol{A}$ and $\Phi$ are not themselves observable; only their gauge-invariant combinations matter.
- **Topological phases**: Line integrals and Wilson loops encode topological information about enclosed fluxes.
- **Experimental access**: Interference, Hall conductance, flux quantization, and SQUID measurements all directly probe gauge-invariant physics.
