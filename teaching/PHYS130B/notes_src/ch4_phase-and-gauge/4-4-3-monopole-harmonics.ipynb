## Lecture Notes

### Wavefunctions in a Monopole Background

When a charged particle moves in the field of a Dirac monopole, the electromagnetic vector potential enters the Schrödinger equation through the **minimal coupling**:

$$ \hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - \frac{q}{c}\boldsymbol{A}$$


where $q$ is the particle's charge and $\boldsymbol{A}$ is the monopole vector potential. For a particle confined to a sphere (or for the angular part of a scattering problem), we must solve the **angular eigenvalue problem** with this modified operator.

The resulting eigenfunctions—**monopole harmonics** $Y_{l,m}^s(\theta,\phi)$—are the generalization of ordinary spherical harmonics to the monopole background.

### The Monopole Quantum Number

The strength of the monopole coupling is characterized by the **monopole quantum number**:

$$ s = \frac{qg}{2\hbar c}$$


In Gaussian units, with $q = e$ (electron charge) and $g$ the monopole charge (in units where Dirac quantization gives $eg = n\hbar c/2$), we have $s = n/2$ (which must be an integer or half-integer, depending on $n$).

For a Dirac monopole (the minimal monopole consistent with one electron), $eg = \hbar c/2$, so $s = 1/2$ for an electron.

:::{admonition} Key Definition: Monopole Quantum Number
:class: important

The monopole quantum number $s = qg/(2\hbar c)$ determines how strongly the charged particle couples to the monopole field. It modifies the allowed angular momentum quantum numbers: instead of $l = 0, 1, 2, \ldots$, we have $l = |s|, |s|+1, |s|+2, \ldots$

This shift in allowed $l$ values is a topological effect of the monopole field.
:::

### Definition and Properties of Monopole Harmonics

**Monopole harmonics** $Y_{l,m}^s(\theta,\phi)$ are defined as eigenfunctions of the **modified orbital angular momentum operators** in the monopole background.

In the monopole background, the canonical angular momentum becomes:

$$ \hat{\boldsymbol{L}} = \boldsymbol{r} \times \left(\hat{\boldsymbol{p}} - \frac{q}{c}\boldsymbol{A}\right)$$


On the sphere (where only angular dependence matters), this can be rewritten as:

$$ \hat{\boldsymbol{L}} = \hat{\boldsymbol{L}}_\text{orb} - s\hat{r}$$


where $\hat{\boldsymbol{L}}_\text{orb}$ is the ordinary orbital angular momentum and the $-s\hat{r}$ term represents the **self-angular momentum of the electromagnetic field** (the spin angular momentum of the electromagnetic field configuration around the monopole).

The monopole harmonics satisfy:

$$ \hat{L}^2 Y_{l,m}^s = \hbar^2 l(l + 1 - 2s) Y_{l,m}^s$$


$$ \hat{L}_z Y_{l,m}^s = \hbar m Y_{l,m}^s$$


with quantum numbers:

- $l = |s|, |s|+1, |s|+2, \ldots$ (orbital angular momentum quantum number, shifted!)
- $m = -l, -l+1, \ldots, l$ (magnetic quantum number, as usual)
- $s$ = monopole quantum number (fixed parameter of the problem)

### Comparison with Ordinary Spherical Harmonics

**For $s = 0$ (no monopole)**:
- $l = 0, 1, 2, \ldots$ (all non-negative integers)
- Eigenvalues: $\hat{L}^2 Y_l^m = \hbar^2 l(l+1) Y_l^m$
- Familiar spherical harmonics

**For $s \neq 0$ (with monopole)**:
- $l = |s|, |s|+1, |s|+2, \ldots$ (shifted!)
- Eigenvalues: $\hat{L}^2 Y_{l,m}^s = \hbar^2 l(l + 1 - 2s) Y_{l,m}^s$ (modified!)
- Monopole harmonics

:::{admonition} Physical Interpretation: Angular Momentum Shift
:class: note

The shift from $l(l+1)$ to $l(l+1-2s)$ reflects that the monopole field carries angular momentum. A particle in this field must have its canonical angular momentum modified by the field's angular momentum contribution.

For $s = 1/2$ (an electron in a Dirac monopole background), the minimum allowed orbital angular momentum is $l = 1/2$. This **half-integer angular momentum** arises from the topological properties of the monopole field, not from internal spin!
:::

### Special Case: Spin-1/2 from Monopole Harmonics

Consider an electron ($q = e$) in a Dirac monopole field with charge $g = \hbar c/(2e)$, giving:

$$ s = \frac{eg}{2\hbar c} = \frac{1}{2}$$


The allowed angular momentum values are:

$$ l = \frac{1}{2}, \frac{3}{2}, \frac{5}{2}, \ldots$$


For the lowest value, $l = 1/2$, the magnetic quantum number ranges as:

$$ m = -\frac{1}{2}, \frac{1}{2}$$


giving **two states**. These two states form a **doublet**, exactly like spin-1/2.

:::{admonition} Deep Insight: Spin and Monopole Fields
:class: example

This result suggests a profound unification: **intrinsic spin-1/2 can be understood as orbital angular momentum in the background of a magnetic charge configuration.**

While the electron does not literally orbit a monopole, topologically the electron's quantum numbers are equivalent to those of a spinor. This perspective hints that spin might arise from a more fundamental gauge theory with monopole-like objects.
:::

### Periodicity and Single-Valuedness

Monopole harmonics satisfy a **non-trivial periodicity condition**:

$$ Y_{l,m}^s(\theta, \phi + 2\pi) = \mathrm{e}^{2\pi \mathrm{i} s} Y_{l,m}^s(\theta, \phi)$$


For the wavefunction to be single-valued on the full sphere ($\phi \in [0, 2\pi]$), we need:

$$ \mathrm{e}^{2\pi \mathrm{i} s} = 1 \quad \text{or} \quad s \in \mathbb{Z}$$


for integer $s$. However, if $s$ is a **half-integer**, then $\mathrm{e}^{2\pi \mathrm{i} s} = -1$, and the wavefunction picks up a minus sign under $2\pi$ rotation. This is acceptable in quantum mechanics—it corresponds to a **projective representation** of the rotation group $SO(3)$.

:::{admonition} Topological Classification: Integer vs. Half-Integer
:class: important

- **Integer $s$**: Ordinary functions on the sphere. Single-valued.
- **Half-integer $s$**: Spinor-like functions. Acquire a minus sign under $2\pi$ rotation; single-valued only as sections of a spin bundle.

This distinction explains why fermions (half-integer spin) and bosons (integer spin) have fundamentally different statistics.
:::

### Orthonormality and Completeness

Monopole harmonics form an **orthonormal basis**:

$$ \int \sin\theta \, \mathrm{d}\theta \, \mathrm{d}\phi \, (Y_{l,m}^s)^* Y_{l',m'}^s = \delta_{ll'}\delta_{mm'}$$


and are **complete**:

$$ \sum_{l=|s|}^\infty \sum_{m=-l}^l Y_{l,m}^s(\theta,\phi) Y_{l,m}^{s*}(\theta',\phi') = \delta(\cos\theta - \cos\theta')\delta(\phi - \phi')$$


This allows any function on the sphere to be expanded as:

$$ f(\theta,\phi) = \sum_{l=|s|}^\infty \sum_{m=-l}^l c_{l,m} Y_{l,m}^s(\theta,\phi)$$


### Connection to Topology and the Hopf Fibration

The monopole field creates a **non-trivial fiber bundle** over the 2-sphere $S^2$. The structure group is $U(1)$ (the phase of the wavefunction), and the bundle is characterized by its **first Chern number**:

$$ c_1 = \frac{s}{\pi} = \frac{qg}{2\pi\hbar c}$$


For a Dirac monopole with $qg = n\hbar c/2$, the Chern number is $c_1 = n/2$. The monopole harmonics are sections of this bundle.

For $s = 1/2$ (corresponding to $c_1 = 1/2$, or equivalently, to a bundle with winding number 1 when restricted to certain circles), the monopole harmonics encode the **Hopf fibration** geometry. This is why spin-1/2 particles have a two-fold rotation symmetry in their angular structure.

### Applications and Physical Relevance

**Scattering theory**: Monopole harmonics are essential for computing scattering amplitudes of electrons off monopoles, predicting observable angular distributions.

**Topological Field Theory**: In gauge theories on spaces with non-trivial bundles, monopole harmonics arise naturally.

**Condensed Matter**: In certain materials (spin ice, pyrochlore lattices), monopole-like excitations emerge as quasiparticles, and their coupling to charge carriers can be described using monopole harmonics.

**Quantum Information**: The spin-1/2 encoding in monopole harmonics has implications for quantum computing and the geometric phases of qubits.
