## Lecture Notes

### Overview

When a charged particle moves on a sphere with a magnetic monopole at the center, a key subtlety arises: neither the canonical angular momentum $\hat{\boldsymbol{L}}_{\rm can} = \boldsymbol{r} \times \hat{\boldsymbol{p}}$ nor the kinetic angular momentum $\hat{\boldsymbol{\Lambda}} = \boldsymbol{r} \times \hat{\boldsymbol{\pi}}$ (where $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - (q/c)\boldsymbol{A}$) is separately conserved. The conserved quantity is the **total angular momentum** $\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} + s\hat{\boldsymbol{r}}$, including the electromagnetic
field's angular momentum $s\hat{\boldsymbol{r}}$. This $\hat{\boldsymbol{J}}$ satisfies the standard SU(2) algebra $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$, but the minimum eigenvalue is shifted: $j \geq |s|$ instead of $j \geq 0$. The eigenfunctions are **monopole harmonics** $Y^s_{j,m}(\theta,\phi)$. The punchline: for $s = 1/2$ (a Dirac monopole), the lowest harmonics form a doublet with $j = 1/2$ \textemdash they *are* spin-1/2 spinors. Half-integer angular momentum is not an independent axiom; it emerges from gauge topology.

### Monopole Quantum Number

:::{admonition} Monopole Quantum Number
:class: important

The **monopole quantum number** characterizes the coupling strength:

$$
s = \frac{qg}{2\hbar c}
$$ (eq-monopole-qnum)

The Dirac quantization condition $qg = n\hbar c/2$ (from the two-patch construction of §4.4.2) restricts $s$ to integer or half-integer values:

$$
s \in \left\{0,\, \pm\tfrac{1}{2},\, \pm 1,\, \pm\tfrac{3}{2},\, \ldots\right\}
$$

For an electron in the field of a minimal Dirac monopole: $|s| = 1/2$.
:::

### Total Angular Momentum

In the monopole background, the kinetic angular momentum $\hat{\boldsymbol{\Lambda}}$ does not commute with itself in the standard way because the gauge field $\boldsymbol{A}$ depends on position. The physically conserved quantity is:

:::{admonition} Total Angular Momentum in a Monopole Field
:class: important

$$
\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} + s\hat{\boldsymbol{r}}
$$

where $\hat{\boldsymbol{\Lambda}} = \boldsymbol{r} \times \hat{\boldsymbol{\pi}}$ is the kinetic angular momentum and $s\hat{\boldsymbol{r}}$ is the angular momentum carried by the electromagnetic field.

$\hat{\boldsymbol{J}}$ satisfies the **standard** SU(2) commutation relations:

$$
[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\,\epsilon_{ijk}\hat{J}_k
$$

with eigenvalues:

$$
\hat{J}^2 \vert s, j, m\rangle = \hbar^2 j(j+1) \vert s, j, m\rangle, \quad \hat{J}_z \vert s, j, m\rangle = \hbar m \vert s, j, m\rangle
$$

**The monopole modifies only the allowed values of $j$:**

$$
j = |s|,\; |s|+1,\; |s|+2,\; \ldots \quad (\text{not } j = 0, 1, 2, \ldots)
$$
:::

The shift $j_{\min} = |s|$ has a direct physical interpretation: the electromagnetic field already carries angular momentum $s\hat{\boldsymbol{r}}$. The total angular momentum cannot be less than this field contribution.

:::{admonition} Derivation: Minimum Angular Momentum $j_{\min} = |s|$
:class: dropdown information

**From the covariant Laplacian eigenvalue problem.** The Hamiltonian on the sphere is $\hat{H} = -\frac{\hbar^2}{2I}\boldsymbol{D}_\perp^2 + \text{const}$, where the covariant Laplacian in the monopole gauge yields the eigenvalue equation (substituting $\psi = \mathrm{e}^{\mathrm{i}(m-s)\phi}\Theta(\theta)$ and $x = \cos\theta$):

$$
\left[(1-x^2)\partial_x^2 - 2x\,\partial_x - \frac{(m - sx)^2}{1-x^2} + \lambda\right]\Theta = 0
$$

**Step 1: Factor out the endpoint behavior.** Write $\Theta(x) = (1-x)^{(s-m)/2}(1+x)^{(s+m)/2}P(x)$. Substituting and simplifying (the singular $(1-x^2)^{-1}$ terms cancel), $P(x)$ satisfies:

$$
(1-x^2)P'' + [2m - 2(1+s)x]P' + (\lambda - s)P = 0
$$

**Step 2: Identify with the Jacobi equation.** The Jacobi polynomial $P_n^{(a,b)}(x)$ satisfies $(1-x^2)P'' + [b - a - (a+b+2)x]P' + n(n+a+b+1)P = 0$. Matching coefficients:

$$
a = s - m, \quad b = s + m, \quad \lambda - s = n(n + 2s + 1)
$$

For $P_n^{(a,b)}$ to be a well-defined polynomial, we need $n \in \{0,1,2,\ldots\}$ and $n + a, n + b \geq 0$ (i.e., $a, b \geq -n$, which is automatic for the physical range $|m| \leq j$).

**Step 3: Introduce $j = s + n$.** Then:

$$
\lambda = n^2 + (2s+1)n + s = (s+n)(s+n+1) - s^2 = j(j+1) - s^2
$$

Since $n \geq 0$, we have $j \geq s$. For general monopole charge, $j \geq |s|$. The minimum value is:

$$
j_{\min} = |s|, \quad \text{with degeneracy } 2|s| + 1
$$

For $s = 1/2$: $j_{\min} = 1/2$, giving a **doublet** $(m = \pm 1/2)$ — this is the spin-1/2 representation.
:::

:::{admonition} Derivation: Angular-Momentum Commutator
:class: dropdown information

We need to show that $\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} + s\hbar\hat{\boldsymbol{r}}$ satisfies the SU(2) algebra.

**Step 1: Kinetic angular momentum commutator.** The kinetic angular momentum $\hat{\boldsymbol{\Lambda}} = \boldsymbol{r} \times \hat{\boldsymbol{\pi}}$ with $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - (q/c)\boldsymbol{A}$ does *not* close into itself. The kinetic momentum components satisfy $[\hat{\pi}_i, \hat{\pi}_j] = \mathrm{i}\hbar(q/c)\epsilon_{ijk}B_k$ (from §4.1.2). For a monopole field $\boldsymbol{B} = g\hat{\boldsymbol{r}}/r^2$, this gives:

$$
[\hat{\Lambda}_i, \hat{\Lambda}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k - \mathrm{i}\hbar s\hbar\,\epsilon_{ijk}\hat{r}_k
$$

where $s = qg/(2\hbar c)$. The extra term $-\mathrm{i}\hbar s\hbar\,\epsilon_{ijk}\hat{r}_k$ arises from the monopole curvature.

**Step 2: Cross-commutators.** Since $\hat{\boldsymbol{\Lambda}}$ generates rotations of $\hat{\boldsymbol{r}}$:

$$
[\hat{\Lambda}_i, \hat{r}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{r}_k
$$

Also $[\hat{r}_i, \hat{r}_j] = 0$ trivially.

**Step 3: Full commutator.** Expanding $[\hat{J}_i, \hat{J}_j] = [\hat{\Lambda}_i + s\hbar\hat{r}_i,\, \hat{\Lambda}_j + s\hbar\hat{r}_j]$:

$$
= [\hat{\Lambda}_i, \hat{\Lambda}_j] + s\hbar[\hat{\Lambda}_i, \hat{r}_j] - s\hbar[\hat{\Lambda}_j, \hat{r}_i] + s^2\hbar^2\underbrace{[\hat{r}_i, \hat{r}_j]}_{=\,0}
$$

$$
= \bigl(\mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k - \mathrm{i}\hbar s\hbar\,\epsilon_{ijk}\hat{r}_k\bigr) + s\hbar\,\mathrm{i}\hbar\epsilon_{ijk}\hat{r}_k + s\hbar\,\mathrm{i}\hbar\epsilon_{ijk}\hat{r}_k
$$

Wait — let us be careful with the second cross-commutator: $-s\hbar[\hat{\Lambda}_j, \hat{r}_i] = -s\hbar\,\mathrm{i}\hbar\epsilon_{jik}\hat{r}_k = +s\hbar\,\mathrm{i}\hbar\epsilon_{ijk}\hat{r}_k$.

So we get:

$$
= \mathrm{i}\hbar\epsilon_{ijk}\hat{\Lambda}_k \underbrace{- \mathrm{i}\hbar s\hbar\,\epsilon_{ijk}\hat{r}_k + 2\,\mathrm{i}\hbar s\hbar\,\epsilon_{ijk}\hat{r}_k}_{= +\mathrm{i}\hbar s\hbar\,\epsilon_{ijk}\hat{r}_k}
$$

$$
= \mathrm{i}\hbar\epsilon_{ijk}(\hat{\Lambda}_k + s\hbar\hat{r}_k) = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k \quad \checkmark
$$

The field angular momentum term $s\hbar\hat{\boldsymbol{r}}$ precisely compensates the anomalous commutator of $\hat{\boldsymbol{\Lambda}}$.
:::

### Monopole Harmonics

The eigenfunctions of $\hat{J}^2$ and $\hat{J}_z$ in the monopole background are **monopole harmonics**:

:::{admonition} Monopole Harmonics
:class: important

$Y^s_{j,m}(\theta,\phi)$ are simultaneous eigenfunctions of $\hat{J}^2$ and $\hat{J}_z$:

$$
\hat{J}^2 Y^s_{j,m} = \hbar^2 j(j+1) Y^s_{j,m}, \quad \hat{J}_z Y^s_{j,m} = \hbar m Y^s_{j,m}
$$

with $j = |s|, |s|+1, \ldots$ and $m = -j, \ldots, j$.

**Orthonormality:** $\displaystyle\int \sin\theta\,\mathrm{d}\theta\,\mathrm{d}\phi\,(Y^s_{j,m})^* Y^s_{j',m'} = \delta_{jj'}\delta_{mm'}$

**Reduction:** For $s = 0$, $Y^0_{l,m} = Y_{lm}(\theta,\phi)$ (ordinary spherical harmonics).
:::

The explicit form of $Y^s_{j,m}$ requires the two-patch construction (§4.4.2): the monopole gauge field cannot be described by a single vector potential on the whole sphere, so northern and southern hemispheres use separate gauges, related by a transition function on the equator.

Summary of how the monopole modifies the spectrum:

| $s$ | $j_{\min}$ | Degeneracy at $j_{\min}$ |
|-----|-----------|--------------------------|
| $0$ (no monopole) | $0$ | $1$ (singlet) |
| $1/2$ (Dirac) | $1/2$ | $2$ (doublet = spin-1/2) |
| $1$ | $1$ | $3$ (triplet) |
| $3/2$ | $3/2$ | $4$ (quartet) |

:::{admonition} Derivation: Explicit Lowest Monopole Harmonics for $s = 1/2$
:class: dropdown information

For $s = 1/2$, the lowest monopole harmonics have $j = 1/2$ and $m = \pm 1/2$. We construct them explicitly in the northern and southern patches.

**Setup.** In the northern patch (angles $0 \leq \theta < \pi$, avoiding the south pole), the Wu-Yang gauge gives $A_\phi^N = \frac{s\hbar}{r}(\cos\theta - 1)/\sin\theta = \frac{\hbar}{2r}(\cos\theta - 1)/\sin\theta$.

The $\hat{J}_z$ eigenvalue condition $\hat{J}_z Y = \hbar m Y$ with $\hat{J}_z = -\mathrm{i}\hbar\partial_\phi + s\hbar\hat{r}_z$ gives:

$$
\left(-\mathrm{i}\hbar\partial_\phi + s\hbar\cos\theta\right) Y^s_{j,m} = \hbar m\, Y^s_{j,m}
$$

This forces $Y^s_{j,m} \propto \mathrm{e}^{\mathrm{i}(m-s)\phi}\,f(\theta)$. For $s = 1/2$, $m = +1/2$: $Y^{1/2}_{1/2,+1/2} \propto f(\theta)$ (no $\phi$ dependence). For $m = -1/2$: $Y^{1/2}_{1/2,-1/2} \propto \mathrm{e}^{-\mathrm{i}\phi} g(\theta)$.

**Solving for $f(\theta)$ and $g(\theta)$.** The eigenvalue equation $\hat{J}^2 Y = \hbar^2 \cdot \frac{3}{4} Y$ in the monopole background (substituting $j = 1/2$, $j(j+1) = 3/4$) reduces to:

$$
-\frac{1}{\sin\theta}\frac{\mathrm{d}}{\mathrm{d}\theta}\left(\sin\theta\frac{\mathrm{d}f}{\mathrm{d}\theta}\right) + \frac{s^2\cos^2\theta}{\sin^2\theta}f = \left(\frac{3}{4} - s^2\right) f
$$

For $s = 1/2$: with the ansatz $f(\theta) = \cos(\theta/2)$ one verifies $\hat{J}^2 f = \frac{3}{4}\hbar^2 f$. Similarly, $g(\theta) = \sin(\theta/2)$ satisfies the $m = -1/2$ equation.

**Result (normalized).** The lowest monopole harmonics in the north patch are:

$$
Y^{1/2}_{1/2,+1/2}(\theta,\phi) = \frac{1}{\sqrt{4\pi}}\cos\frac{\theta}{2}, \qquad Y^{1/2}_{1/2,-1/2}(\theta,\phi) = \frac{1}{\sqrt{4\pi}}\mathrm{e}^{-\mathrm{i}\phi}\sin\frac{\theta}{2}
$$

**Identification with spin coherent state.** These are exactly the components of the spin-1/2 coherent state $\vert\hat{\boldsymbol{n}}\rangle = \cos(\theta/2)\vert\uparrow\rangle + \mathrm{e}^{\mathrm{i}\phi}\sin(\theta/2)\vert\downarrow\rangle$ up to conventions. The doublet $(Y^{1/2}_{1/2,+1/2}, Y^{1/2}_{1/2,-1/2})$ transforms as a spinor under rotations: a $2\pi$ rotation sends $\theta/2 \to \theta/2 + \pi$, giving an overall factor $\mathrm{e}^{\mathrm{i}\pi} = -1$ (the spinor sign flip).

This confirms the central punchline: **half-integer spin is not an axiom** — it is forced by the topology of the Dirac monopole gauge bundle.
:::


### Punchline: Spin-1/2 from Monopole Topology

:::{admonition} Spin-1/2 Emerges from $s = 1/2$
:class: important

For a Dirac monopole ($|s| = 1/2$), the minimum angular momentum is $j_{\min} = 1/2$. The lowest states are:

$$
\vert \tfrac{1}{2}, +\tfrac{1}{2}\rangle \quad \text{and} \quad \vert \tfrac{1}{2}, -\tfrac{1}{2}\rangle
$$

These form a **doublet** transforming exactly as spin-1/2: under a $2\pi$ rotation, each state acquires a factor $\mathrm{e}^{2\pi\mathrm{i}s} = \mathrm{e}^{\pi\mathrm{i}} = -1$ (the spinor sign flip).

**Conclusion**: Half-integer angular momentum is **not** an independent axiom. It is the minimum angular momentum enforced by the topology of the monopole gauge bundle.
:::

### Connection to Berry Phase

The monopole structure of angular momentum has a direct physical manifestation through Berry phase (§4.2). For a spin-1/2 particle in a slowly rotating magnetic field $\boldsymbol{B}(t)$, as $\boldsymbol{B}$ traces a closed loop on the unit sphere in parameter space, the spin acquires a geometric phase $\gamma = -\frac{1}{2}\Omega$ where $\Omega$ is the solid angle subtended. The Berry curvature is:

$$
\boldsymbol{F} = \frac{1}{2}\frac{\hat{\boldsymbol{n}}}{|\boldsymbol{n}|^2}
$$

:::{admonition} Spin = Monopole in Parameter Space
:class: important

The Berry curvature of a spin-$s$ system is the field of a **monopole of charge $s$** at the origin of parameter space. This is not an analogy \textemdash it is the same mathematical structure.

The Chern number of the Berry connection over any closed surface enclosing the degeneracy is:

$$
c_1 = \frac{1}{2\pi}\oint \boldsymbol{F}\cdot\mathrm{d}\boldsymbol{S} = 2s
$$

This closes the circle: spin-1/2 emerges from monopole topology (§4.4.2), and the Berry phase of spin-1/2 is itself a monopole field (§4.2.1).
:::

:::{admonition} Derivation: Berry Curvature of Spin-$s$ as a Monopole Field
:class: dropdown information

Consider a spin-$s$ particle in a magnetic field $\boldsymbol{B} = B\hat{\boldsymbol{n}}$, where $\hat{\boldsymbol{n}} = (\sin\theta\cos\phi, \sin\theta\sin\phi, \cos\theta)$ is a parameter slowly varied. The Hamiltonian is $\hat{H} = -\gamma B\,\hat{\boldsymbol{S}} \cdot \hat{\boldsymbol{n}}$, with ground state $\vert s, m{=}{-s}; \hat{\boldsymbol{n}}\rangle$ (energy eigenstate aligned with $\hat{\boldsymbol{n}}$).

**Step 1: Berry connection.** Using the spin-$s$ coherent state $\vert \hat{\boldsymbol{n}}\rangle = \mathrm{e}^{-\mathrm{i}\phi \hat{S}_z/\hbar}\mathrm{e}^{-\mathrm{i}\theta \hat{S}_y/\hbar}\vert s, -s\rangle$ (in the Wu-Yang gauge):

$$
A_\theta = \mathrm{i}\hbar\langle \hat{\boldsymbol{n}} \vert \partial_\theta \vert \hat{\boldsymbol{n}}\rangle = 0
$$

$$
A_\phi = \mathrm{i}\hbar\langle \hat{\boldsymbol{n}} \vert \partial_\phi \vert \hat{\boldsymbol{n}}\rangle = s\hbar(\cos\theta - 1)
$$

(This parallels the Berry connection computation for spin-1/2 in §4.2.1, generalized to spin-$s$.)

**Step 2: Berry curvature.** The curvature 2-form on the sphere is:

$$
F_{\theta\phi} = \partial_\theta A_\phi - \partial_\phi A_\theta = s\hbar\frac{\partial}{\partial\theta}(\cos\theta - 1) = -s\hbar\sin\theta
$$

Converting to a radial field in parameter space (with $r = 1$ on the unit sphere):

$$
\boldsymbol{F} = \frac{F_{\theta\phi}}{r^2\sin\theta}\hat{\boldsymbol{r}} = \frac{-s\hbar\sin\theta}{\sin\theta}\hat{\boldsymbol{r}} = -s\hbar\,\hat{\boldsymbol{r}}
$$

This is exactly the field of a **monopole of charge $s$** at the origin of parameter space.

**Step 3: Chern number.** Integrating the Berry curvature over the full sphere:

$$
c_1 = \frac{1}{2\pi\hbar}\oint \boldsymbol{F} \cdot \mathrm{d}\boldsymbol{S} = \frac{1}{2\pi\hbar}\int_0^{2\pi}\mathrm{d}\phi\int_0^\pi (-s\hbar)\sin\theta\,\mathrm{d}\theta = \frac{-s\hbar \cdot 4\pi}{2\pi\hbar} = -2s
$$

The magnitude $|c_1| = 2|s|$ is an integer (by the Dirac quantization condition), confirming the topological nature of spin.
:::


:::{admonition} Derivation: North and South Patch Potentials and Dirac Quantization
:class: dropdown information

The vector potential of a Dirac monopole cannot be defined globally — it requires (at least) two patches with a transition function on the overlap. We derive the Dirac quantization condition by demanding single-valuedness of the wavefunction.

**Monopole field.** A magnetic monopole of strength $g$ at the origin has

$$
\boldsymbol{B} = \frac{g}{4\pi}\frac{\hat{\boldsymbol{r}}}{r^2}.
$$

**North and South patch potentials.** In spherical coordinates $(r, \theta, \phi)$, the two-patch decomposition is:

$$
\boldsymbol{A}^{(N)} = \frac{g}{4\pi r}\frac{1 - \cos\theta}{\sin\theta}\hat{\boldsymbol{\phi}}, \quad (\text{regular everywhere except } \theta = \pi)
$$

$$
\boldsymbol{A}^{(S)} = -\frac{g}{4\pi r}\frac{1 + \cos\theta}{\sin\theta}\hat{\boldsymbol{\phi}}, \quad (\text{regular everywhere except } \theta = 0).
$$

Both give the same $\boldsymbol{B} = \nabla\times\boldsymbol{A}$ in their respective domains. On the equatorial overlap ($\theta = \pi/2$), they differ by a gauge transformation:

$$
\boldsymbol{A}^{(N)} - \boldsymbol{A}^{(S)} = \frac{g}{2\pi r\sin\theta}\hat{\boldsymbol{\phi}} = \boldsymbol{\nabla}\left(\frac{g\phi}{2\pi}\right).
$$

The transition function is $\chi(\phi) = \frac{g\phi}{2\pi}$.

**Wavefunction matching.** A charged particle (charge $q$) described by $\psi^{(N)}$ in the north patch and $\psi^{(S)}$ in the south must satisfy

$$
\psi^{(N)} = \mathrm{e}^{\mathrm{i}q\chi/\hbar}\psi^{(S)} = \mathrm{e}^{\mathrm{i}qg\phi/(2\pi\hbar)}\psi^{(S)}.
$$

**Quantization.** For $\psi^{(N)}$ to be single-valued around the equator ($\phi \to \phi + 2\pi$), the phase $\mathrm{e}^{\mathrm{i}qg/\hbar}$ must equal 1:

$$
\frac{qg}{\hbar} = 2\pi n, \quad n \in \mathbb{Z} \quad \Longrightarrow \quad qg = n \cdot 2\pi\hbar = n\, h.
$$

This is the **Dirac quantization condition** $qg = nh/2\pi$ — derived directly from the single-valuedness of quantum states. The monopole charge $g$ is quantized in units of $h/q$, and the monopole quantum number is $s = qg/(4\pi\hbar) \in \tfrac{1}{2}\mathbb{Z}$.
:::

### Summary

- The **conserved total angular momentum** in a monopole field is $\hat{\boldsymbol{J}} = \hat{\boldsymbol{\Lambda}} + s\hat{\boldsymbol{r}}$, satisfying standard SU(2): $[\hat{J}_i, \hat{J}_j] = \mathrm{i}\hbar\epsilon_{ijk}\hat{J}_k$.
- **Monopole harmonics** $Y^s_{j,m}$ are eigenstates of $\hat{J}^2$ and $\hat{J}_z$ with eigenvalues $\hbar^2 j(j+1)$, but $j_{\min} = |s|$ instead of 0.
- The **shift $j_{\min} = |s|$** reflects the angular momentum already carried by the electromagnetic field.
- **For $s = 1/2$** (Dirac monopole): the lowest harmonics form a spin-1/2 doublet \textemdash half-integer angular momentum emerges from topology, not as an axiom.
- **Berry connection**: the Berry curvature of spin-$s$ is a monopole of charge $s$ in parameter space \textemdash spin is topological.
