## Lecture Notes

### Overview

The classical motion of a charged particle in a uniform magnetic field leads to cyclotron orbits—circular trajectories with a characteristic frequency independent of energy. This classical picture provides crucial intuition but faces a profound puzzle: under strong magnetic fields at low temperatures, the Hall conductance "locks in" to discrete quantized values with parts-per-billion precision. This quantization cannot be explained classically and demands quantum mechanics. Understanding cyclotron motion and the magnetic length scale $\ell_B$ provides the foundation for Landau quantization and the quantum Hall effects—among the most important phenomena in condensed-matter physics.

### Classical Cyclotron Motion

**Equations of Motion.**

A charged particle with charge $q$ and mass $m$ in a uniform magnetic field $\boldsymbol{B} = B\hat{z}$ experiences the Lorentz force:

$$
m\boldsymbol{a} = q\boldsymbol{v} \times \boldsymbol{B}
$$

In Cartesian components:

$$
m\dot{v}_x = qv_yB, \quad m\dot{v}_y = -qv_xB, \quad m\dot{v}_z = 0
$$

Motion parallel to $\boldsymbol{B}$ is unaffected. For the perpendicular components:

$$
\ddot{v}_x + \omega_c^2 v_x = 0, \quad \ddot{v}_y + \omega_c^2 v_y = 0
$$

where:

:::{admonition} Cyclotron Frequency
:class: important

$$
\omega_c = \frac{qB}{m}
$$

**Key properties:**
- Independent of particle velocity or kinetic energy
- Depends only on charge-to-mass ratio and field strength
- Orbital period: $T_c = 2\pi/\omega_c = 2\pi m/(qB)$ is also energy-independent
- For electrons: $\omega_c = eB/m_e$ (note: $e < 0$ in SI; the sign is built into $q$)

:::

The general solution describes circular motion in the $xy$-plane:

$$
v_x(t) = v_\perp \cos(\omega_c t + \phi_0), \quad v_y(t) = v_\perp \sin(\omega_c t + \phi_0)
$$

where $v_\perp = \sqrt{v_x^2 + v_y^2}$ is the speed perpendicular to $\boldsymbol{B}$ and $\phi_0$ is an arbitrary phase. Integrating to find position:

$$
x(t) = x_0 + \frac{v_\perp}{\omega_c}\sin(\omega_c t + \phi_0), \quad y(t) = y_0 - \frac{v_\perp}{\omega_c}\cos(\omega_c t + \phi_0)
$$

This is a circular orbit centered at the **guiding center** $(x_0, y_0)$.

**Cyclotron Radius and Orbital Period.**

:::{admonition} Cyclotron Radius (Larmor Radius)
:class: important

The radius of the circular orbit is:

$$
r_c = \frac{v_\perp}{\omega_c} = \frac{mv_\perp}{qB}
$$

Stronger fields confine orbits more tightly; radius scales with perpendicular momentum.

:::

:::{admonition} Orbital Period: Energy-Independent Timescale
:class: note

The time for one complete orbit is:

$$
T_c = \frac{2\pi}{\omega_c} = \frac{2\pi m}{qB}
$$

Remarkably, $T_c$ is completely independent of the particle's kinetic energy. A slow particle and a fast particle orbit in circles of different radii, but both take the same time to complete one loop. This is why cyclotron accelerators work: by applying a radiofrequency voltage synchronized to $f_c = \omega_c/(2\pi)$, particles are accelerated in phase regardless of their instantaneous energy.

:::

### Classical Hall Effect

When charge carriers flow through a conductor in a perpendicular magnetic field, they are deflected by the Lorentz force and accumulate on one edge until an electric field balances the magnetic force. This produces a Hall voltage across the conductor.

For a conductor with $n$ charge carriers per unit volume and carrier charge $q$:

:::{admonition} Classical Hall Conductance
:class: important

$$
\sigma_{xy} = \frac{nq}{B}
$$

Equivalently, the Hall coefficient is $R_H = 1/(nq)$, and the Hall resistivity is $\rho_H = B/(nq)$. Classically, Hall conductance varies continuously with $B$.

:::

### Integer Quantum Hall Effect: A Window to Quantization

At low temperature (typically below 4 K) and strong magnetic field (several Tesla), experiments reveal something startling that classical physics cannot explain:

:::{admonition} Integer Quantum Hall Effect
:class: important

The Hall conductance becomes quantized in integer units of $e^2/h$:

$$
\sigma_{xy} = \nu \frac{e^2}{h}
$$

where $\nu = 1, 2, 3, \ldots$ is the filling factor (number of occupied Landau levels).

**Key features:**
- Quantization is exact to parts per billion—better than any classical physical parameter
- Conductance is independent of sample details (geometry, disorder, purity)
- Hall resistance shows plateaus at $R_{xy} = h/(\nu e^2)$
- Simultaneously, longitudinal resistance $R_{xx}$ drops to nearly zero on the plateaus
- This precision (von Klitzing, 1980) is now used to define electrical resistance standards

:::

:::{admonition} The Quantum Puzzle
:class: attention

Classical Hall effect predicts $\sigma_{xy} = nq/B$, which is continuous in $B$. But experiments show discrete plateaus. Explanation requires:

1. The magnetic field quantizes electron motion into Landau levels (discrete energy states)
2. Each Landau level can hold a fixed number of electrons (set by its degeneracy)
3. When the Fermi level lies between Landau levels, scattering is suppressed
4. This explains both the quantization and extraordinary precision

These ideas are developed in **§4.3.2 Landau Quantization**.

:::

### Magnetic Length: The Quantum Scale

In a strong magnetic field, the quantum mechanical extent of a particle's wavefunction is set by the **magnetic length**:

:::{admonition} Magnetic Length
:class: important

$$
\ell_B = \sqrt{\frac{\hbar}{qB}}
$$

**Physical interpretation:**
- Characteristic size of a quantum wavefunction in a magnetic field
- Plays the role of lattice constant in condensed-matter systems
- Independent of particle mass; depends only on charge and field strength
- For electrons in a 1 T field: $\ell_B \approx 26$ nm
- For electrons in a 10 T field: $\ell_B \approx 8$ nm

:::

The magnetic length represents the minimum uncertainty in position allowed by the uncertainty principle in a magnetic field. In the lowest Landau level, commutation relations between canonical momentum $\boldsymbol{\pi} = \boldsymbol{p} - q\boldsymbol{A}$ and position lead to $\Delta x \, \Delta y \sim \hbar / (qB)$. With equal uncertainty in both perpendicular directions, this yields $\ell_B = \sqrt{\hbar/(qB)}$. A detailed quantum mechanical analysis is presented in **§4.3.2**.

### Summary

- **Cyclotron frequency** $\omega_c = qB/m$ and **cyclotron radius** $r_c = mv_\perp/(qB)$ characterize classical orbits; the period $T_c = 2\pi/\omega_c$ is constant independent of energy.
- **Classical Hall effect** arises from Lorentz-force balance: Hall conductance $\sigma_{xy} = nq/B$ is continuous classically.
- **Integer Quantum Hall Effect** quantizes $\sigma_{xy}$ to exact multiples of $e^2/h$—impossible classically, revealing discrete Landau levels.
- **Magnetic length** $\ell_B = \sqrt{\hbar/(qB)}$ sets the quantum wavefunction size scale, emerging from the uncertainty principle.
