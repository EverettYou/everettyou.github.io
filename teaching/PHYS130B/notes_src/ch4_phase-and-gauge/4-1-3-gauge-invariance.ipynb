## Lecture Notes

### Overview

In §4.1.1–4.1.2, we built the gauge-invariant Hamiltonian and derived the electromagnetic fields and forces. This section asks: which quantities in this framework are physical (gauge-invariant) and which are conventional (gauge-dependent)? The answer reveals that gauge invariance is not a symmetry of nature—it is a **redundancy** in our description.

### Gauge Transformations

A **gauge transformation** modifies the potentials and the wavefunction simultaneously, leaving all physics unchanged:

:::{admonition} Gauge Transformation
:class: important

$$
\psi \to \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi, \qquad \boldsymbol{A} \to \boldsymbol{A} + \nabla\alpha, \qquad \Phi \to \Phi - \partial_t\alpha
$$

All three transform in concert: the Schrödinger equation $\mathrm{i}\hbar\partial_t\psi = \hat{H}\psi$ with $\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi$ takes the same form in every gauge.
:::

:::{admonition} Derivation: Gauge Invariance of the Schrödinger Equation
:class: dropdown information

Set $\psi' = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi$. The covariant derivative gives (§4.1.1):

$$
(\hat{\boldsymbol{p}} - q\boldsymbol{A}')\psi' = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}(\hat{\boldsymbol{p}} - q\boldsymbol{A})\psi
$$

Squaring: $(\hat{\boldsymbol{p}} - q\boldsymbol{A}')^2\psi' = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2\psi$. For the time derivative:

$$
\mathrm{i}\hbar\partial_t\psi' = -q(\partial_t\alpha)\mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi + \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\mathrm{i}\hbar\partial_t\psi
$$

The $-q\partial_t\alpha$ term is compensated by $\Phi \to \Phi - \partial_t\alpha$ in the Hamiltonian.
:::

### Momentum: Canonical, Kinetic, and Field

The gauge-invariant Hamiltonian $\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + q\Phi$ gives rise to three distinct momentum quantities:

:::{admonition} Three Types of Momentum
:class: important

| | Canonical | Kinetic | Field |
|---|---|---|---|
| **Symbol** | $\hat{\boldsymbol{p}}$ | $\hat{\boldsymbol{\pi}}$ | $q\boldsymbol{A}$ |
| **Operator** | $-\mathrm{i}\hbar\nabla$ | $\hat{\boldsymbol{p}} - q\boldsymbol{A}$ | $q\boldsymbol{A}(\boldsymbol{r})$ |
| **Gauge** | Dependent | Invariant | Dependent |
| **Conserved** | Only if $\partial_i\hat{H} = 0$ | No (Lorentz force) | No |

The relation $\hat{\boldsymbol{p}} = \hat{\boldsymbol{\pi}} + q\boldsymbol{A}$ decomposes canonical momentum into a physical part (kinetic, $= m\boldsymbol{v}$) and a field part.
:::

### Energy: Kinetic, Potential, and Total

The same distinction applies to energy:

:::{admonition} Three Types of Energy
:class: important

| | Kinetic | Potential | Total |
|---|---|---|---|
| **Expression** | $\frac{\hat{\boldsymbol{\pi}}^2}{2m} = \frac{(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2}{2m}$ | $q\Phi$ | $\hat{H} = \frac{\hat{\boldsymbol{\pi}}^2}{2m} + q\Phi$ |
| **Gauge** | Invariant | Dependent | Dependent |
| **Conserved** | No | No | Yes (if $\hat{H}$ is time-independent) |

The kinetic energy $\hat{\boldsymbol{\pi}}^2/2m$ is gauge-invariant (built from gauge-invariant $\hat{\boldsymbol{\pi}}$). The potential energy $q\Phi$ is gauge-dependent (since $\Phi$ shifts). The total energy (Hamiltonian eigenvalue) is gauge-invariant—energy eigenvalues do not depend on gauge choice.
:::

:::{admonition} Caution: Total Hamiltonian vs Energy Eigenvalues
:class: caution

The *operator* $\hat{H}$ changes form under a gauge transformation (because $\boldsymbol{A}$ and $\Phi$ shift), but its *eigenvalues* $E_n$ do not. The gauge-transformed eigenstate $\psi_n' = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi_n$ satisfies $\hat{H}'\psi_n' = E_n\psi_n'$ with the same $E_n$.
:::

### Gauge-Invariant vs Gauge-Dependent Quantities

:::{admonition} Classification of Physical Quantities
:class: important

**Gauge-invariant (physical):**

| Quantity | Symbol |
|----------|--------|
| Electric field | $\boldsymbol{E}$ |
| Magnetic field | $\boldsymbol{B}$ |
| Field strength tensor | $F_{\mu\nu}$ |
| Kinetic momentum | $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ |
| Kinetic energy | $\hat{\boldsymbol{\pi}}^2/2m$ |
| Energy eigenvalues | $E_n$ |
| Probability density | $\vert\psi\vert^2$ |
| Probability current | $\boldsymbol{j} = \frac{1}{m}\mathrm{Re}(\psi^*\hat{\boldsymbol{\pi}}\psi)$ |
| Transition probabilities | $\vert\langle\phi\vert\psi\rangle\vert^2$ |
| Closed-loop phase | $\oint \boldsymbol{A} \cdot \mathrm{d}\boldsymbol{l}$ |

**Gauge-dependent (conventional):**

| Quantity | Symbol |
|----------|--------|
| Vector potential | $\boldsymbol{A}$ |
| Scalar potential | $\Phi$ |
| Canonical momentum | $\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla$ |
| Potential energy | $q\Phi$ |
| Wavefunction phase | $\arg(\psi)$ |
:::

### Gauge Invariance as Redundancy

:::{admonition} Gauge Invariance Is Not a Symmetry
:class: important

A **physical symmetry** (rotation, translation) maps one physical state to a *different* physical state that happens to obey the same laws.

A **gauge transformation** maps one *description* to another description of the **same** physical state. The physics has not changed—only the bookkeeping.

Gauge invariance is therefore a **redundancy**: multiple mathematical descriptions ($\psi$, $\boldsymbol{A}$, $\Phi$ in different gauges) correspond to one physical reality. Requiring gauge invariance is not imposing a new law of nature—it is **projecting out the redundant degrees of freedom** to identify what is physical.
:::

This perspective explains why gauge invariance is so powerful: it constrains the theory not because it adds structure, but because it *removes* unphysical freedom. The minimal coupling prescription, the form of the Hamiltonian, the Lorentz force—all follow from demanding that the redundant description be self-consistent.

### Gauge Fixing

In practice, we choose a gauge to simplify calculations. Common choices:

:::{admonition} Common Gauge Choices
:class: note

- **Coulomb gauge** ($\nabla \cdot \boldsymbol{A} = 0$): Standard for nonrelativistic problems. $\Phi$ becomes the instantaneous Coulomb potential.
- **Lorenz gauge** ($\partial_\mu A^\mu = 0$): Manifestly Lorentz-covariant; preferred in relativistic QFT.
- **Temporal gauge** ($\Phi = 0$): Used in lattice gauge theory.

All gauges give identical physics; the choice is a computational convenience.
:::

### Summary

- **Gauge transformations** act on $\psi$ (phase), $\boldsymbol{A}$ (gradient), and $\Phi$ (time derivative) simultaneously, preserving all physics.
- **Momentum**: canonical $\hat{\boldsymbol{p}}$ (gauge-dependent), kinetic $\hat{\boldsymbol{\pi}}$ (gauge-invariant), field $q\boldsymbol{A}$ (gauge-dependent).
- **Energy**: kinetic $\hat{\boldsymbol{\pi}}^2/2m$ (gauge-invariant), potential $q\Phi$ (gauge-dependent), eigenvalues $E_n$ (gauge-invariant).
- **Gauge invariance is redundancy**, not symmetry: multiple descriptions of the same state, projected to physical reality by requiring consistency.

(c) Construct a gauge-invariant momentum operator from $\hat{\boldsymbol{p}}$ and $\boldsymbol{A}$.

**2. Probability current.** The probability current is $\boldsymbol{j} = \frac{1}{m}\mathrm{Re}(\psi^*\hat{\boldsymbol{\pi}}\psi)$ with $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$.

(a) Show that $\boldsymbol{j}$ is gauge-invariant.

(b) Show that the continuity equation $\partial_t\vert\psi\vert^2 + \nabla \cdot \boldsymbol{j} = 0$ holds in every gauge.

(b) Show that both agree on $\langle\hat{\boldsymbol{\pi}}\rangle$.
