## Lecture Notes

### Overview

The gauge-invariant Hamiltonian $\hat{H} = \frac{(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2}{2m} + q\Phi$, derived in §4.1.1 purely from local U(1) phase symmetry, contains a profound discovery: it automatically describes charged particles interacting with electromagnetic fields. By computing equations of motion, we extract the Lorentz force and identify the electric and magnetic fields. The gauge principle transforms a mathematical symmetry requirement into the fundamental interaction that shapes the universe.

### Kinetic Momentum

The Heisenberg equation for position gives:

$$
\frac{\mathrm{d}\langle\boldsymbol{x}\rangle}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}\langle[\hat{\boldsymbol{x}}, \hat{H}]\rangle = \frac{\langle\hat{\boldsymbol{p}} - q\boldsymbol{A}\rangle}{m}
$$

using $[\hat{x}_i, \hat{p}_j] = \mathrm{i}\hbar\delta_{ij}$ and $[\hat{x}_i, \boldsymbol{A}(\hat{\boldsymbol{x}})] = 0$.

:::{admonition} Kinetic Momentum
:class: important

$$
\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}, \qquad \boldsymbol{v} = \frac{\hat{\boldsymbol{\pi}}}{m}
$$

The **kinetic momentum** $\hat{\boldsymbol{\pi}}$ determines the particle's velocity. The canonical momentum $\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla$ is gauge-dependent; only $\hat{\boldsymbol{\pi}}$ is directly observable.
:::

:::{admonition} Caution: Kinetic Momentum Does Not Commute
:class: caution

Unlike canonical momentum ($[\hat{p}_i, \hat{p}_j] = 0$), kinetic momentum components satisfy $[\hat{\pi}_i, \hat{\pi}_j] = \mathrm{i}q\hbar\epsilon_{ijk}B_k$ in a magnetic field. This noncommutativity is central to Landau quantization (§4.3).
:::

### The Lorentz Force Law

Computing $\frac{\mathrm{d}\langle\hat{\boldsymbol{\pi}}\rangle}{\mathrm{d}t}$ requires the commutator $[\hat{\boldsymbol{p}}, \hat{H}]$. Using $[\hat{p}_i, A_j] = -\mathrm{i}\hbar\partial_i A_j$ and the product rule on $(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2$, one extracts gradient and curl terms from the gauge potentials.

:::{admonition} Lorentz Force from Gauge Theory
:class: important

$$
m\frac{\mathrm{d}\langle\hat{\boldsymbol{v}}\rangle}{\mathrm{d}t} = q\left\langle\boldsymbol{E} + \frac{1}{2}(\hat{\boldsymbol{v}}\times\boldsymbol{B} - \boldsymbol{B}\times\hat{\boldsymbol{v}})\right\rangle
$$

where $\hat{\boldsymbol{v}} = \hat{\boldsymbol{\pi}}/m$, $\boldsymbol{E} = -\nabla\Phi - \frac{\partial\boldsymbol{A}}{\partial t}$, and $\boldsymbol{B} = \nabla \times \boldsymbol{A}$.

This is the **quantum Lorentz force law**, derived entirely from the gauge-invariant Hamiltonian. The magnetic force involves the symmetrized product $\frac{1}{2}(\hat{\boldsymbol{v}}\times\boldsymbol{B} - \boldsymbol{B}\times\hat{\boldsymbol{v}})$ because $\hat{\boldsymbol{v}}$ and $\boldsymbol{B}(\hat{\boldsymbol{x}})$ do not commute.
:::

:::{admonition} Derivation: Equation of Motion for Kinetic Momentum
:class: dropdown information

**Step 1 — Split into canonical momentum and vector potential.**
We want $m\frac{\mathrm{d}\langle \hat{v}_i\rangle}{\mathrm{d}t}$ where $\hat{v}_i = \hat{\pi}_i/m = (\hat{p}_i - qA_i)/m$. Since $A_i(\boldsymbol{x},t)$ depends on both position and time,

$$
\begin{split}
m\frac{\mathrm{d}\langle \hat{v}_i\rangle}{\mathrm{d}t}
&= \frac{\mathrm{d}\langle\hat{p}_i\rangle}{\mathrm{d}t} - q\frac{\mathrm{d}\langle A_i\rangle}{\mathrm{d}t}
\end{split}
$$

We evaluate each term separately.

**Step 2 — Heisenberg equation for canonical momentum.**
$\frac{\mathrm{d}\langle\hat{p}_i\rangle}{\mathrm{d}t} = \frac{1}{\mathrm{i}\hbar}\langle[\hat{p}_i, \hat{H}]\rangle$ with $\hat{H} = \frac{1}{2m}\hat{\pi}_j\hat{\pi}_j + q\Phi$. Since $[\hat{p}_i, \hat{p}_j^2]=0$, the only contributions come from terms involving $A_j$ and $\Phi$.

**Step 3 — Evaluate each commutator** using $[\hat{p}_i, f(\boldsymbol{x})] = -\mathrm{i}\hbar\,\partial_i f$ and the product rule $[A, BC] = [A,B]C + B[A,C]$:

$$
[\hat{p}_i, A_j\hat{p}_j] = [\hat{p}_i, A_j]\hat{p}_j = -\mathrm{i}\hbar(\partial_i A_j)\hat{p}_j
$$

$$
[\hat{p}_i, \hat{p}_j A_j] = \hat{p}_j[\hat{p}_i, A_j] = -\mathrm{i}\hbar\,\hat{p}_j(\partial_i A_j)
$$

$$
[\hat{p}_i, q^2 A_j^2] = -\mathrm{i}\hbar\,q^2\,\partial_i(A_j^2) = -2\mathrm{i}\hbar\,q^2 A_j(\partial_i A_j)
$$

$$
[\hat{p}_i, q\Phi] = -\mathrm{i}\hbar\,q\,\partial_i\Phi
$$

**Step 4 — Collect and factor out $\hat{\pi}_j$.**
Assembling $[\hat{p}_i, \hat{H}]$ from the four commutators above (the cross term in $\hat{\pi}^2 = \hat{p}^2 - q\hat{p}\cdot A - qA\cdot\hat{p} + q^2 A^2$ carries a factor $-q$ on the first two and $+q^2$ on the last):

$$
\begin{split}
\frac{1}{\mathrm{i}\hbar}[\hat{p}_i, \hat{H}]
&= \frac{q}{2m}\bigl[(\partial_i A_j)\hat{p}_j + \hat{p}_j(\partial_i A_j) - 2q A_j(\partial_i A_j)\bigr] - q\,\partial_i\Phi
\end{split}
$$

Recognizing $\hat{p}_j - qA_j = \hat{\pi}_j$ in the bracket:

$$
\begin{split}
\frac{\mathrm{d}\langle\hat{p}_i\rangle}{\mathrm{d}t}
&= \frac{q}{2m}\bigl\langle(\partial_i A_j)\hat{\pi}_j + \hat{\pi}_j(\partial_i A_j)\bigr\rangle - q\langle\partial_i\Phi\rangle
\end{split}
$$

The symmetrized product $(\partial_i A_j)\hat{\pi}_j + \hat{\pi}_j(\partial_i A_j)$ appears because $\hat{\pi}_j$ and $\partial_i A_j$ do not commute.

**Step 5 — Total time derivative of $A_i$.**
By the Heisenberg equation for $A_i(\hat{\boldsymbol{x}}, t)$:

$$
\begin{split}
\frac{\mathrm{d}\langle A_i\rangle}{\mathrm{d}t}
&= \left\langle\frac{\partial A_i}{\partial t}\right\rangle + \frac{1}{2m}\bigl\langle(\partial_j A_i)\hat{\pi}_j + \hat{\pi}_j(\partial_j A_i)\bigr\rangle
\end{split}
$$

The second term is the symmetrized convective derivative $\hat{v}_j\,\partial_j A_i$.

**Step 6 — Combine and identify $\boldsymbol{E}$ and $\boldsymbol{B}$.**
Substituting Steps 4 and 5 into Step 1:

$$
\begin{split}
m\frac{\mathrm{d}\langle \hat{v}_i\rangle}{\mathrm{d}t}
&= -q\left\langle\partial_i\Phi + \frac{\partial A_i}{\partial t}\right\rangle \\
&\quad + \frac{q}{2m}\bigl\langle\hat{\pi}_j(\partial_i A_j - \partial_j A_i) + (\partial_i A_j - \partial_j A_i)\hat{\pi}_j\bigr\rangle
\end{split}
$$

The first line gives $q\langle E_i\rangle$ since $E_i = -\partial_i\Phi - \partial_t A_i$. In the second line, recognize $\partial_i A_j - \partial_j A_i = \epsilon_{ijk}B_k$, so

$$
\begin{split}
m\frac{\mathrm{d}\langle \hat{v}_i\rangle}{\mathrm{d}t}
&= q\langle E_i\rangle + \frac{q}{2m}\,\epsilon_{ijk}\bigl\langle\hat{\pi}_j B_k + B_k\hat{\pi}_j\bigr\rangle \\
&= q\left\langle E_i + \frac{1}{2}(\hat{\boldsymbol{v}}\times\boldsymbol{B} - \boldsymbol{B}\times\hat{\boldsymbol{v}})_i\right\rangle
\end{split}
$$

The magnetic term is **not** a simple cross product: because $\hat{v}_j$ and $B_k$ do not commute, the symmetrized combination $\frac{1}{2}(\hat{\boldsymbol{v}}\times\boldsymbol{B} - \boldsymbol{B}\times\hat{\boldsymbol{v}})$ is the correct quantum-mechanical operator ordering.
:::

:::{admonition} Discussion
:class: dropdown tip

In classical mechanics, the Lorentz force $m\boldsymbol{a} = q(\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B})$ is an empirical law. In quantum mechanics, we derived it from the gauge principle alone. Does this mean quantum mechanics is "deeper" than classical mechanics? Could a different gauge group produce a different force law?
:::

### Electric and Magnetic Fields from Potentials

The gauge potentials $(\Phi, \boldsymbol{A})$ are gauge-dependent and not directly observable. The physical fields are:

$$
\boldsymbol{E} = -\nabla\Phi - \frac{\partial\boldsymbol{A}}{\partial t}, \qquad \boldsymbol{B} = \nabla \times \boldsymbol{A}
$$

In spacetime notation with $A_\mu = (\Phi, \boldsymbol{A})$, both fields are encoded in the **field strength tensor**:

:::{admonition} Field Strength Tensor
:class: important

$$
F_{\mu\nu} = \partial_\mu A_\nu - \partial_\nu A_\mu
$$

$F_{\mu\nu}$ is **gauge-invariant**: under $A_\mu \to A_\mu + \partial_\mu\chi$, the antisymmetry gives $F'_{\mu\nu} = F_{\mu\nu}$ because $\partial_\mu\partial_\nu\chi = \partial_\nu\partial_\mu\chi$.
:::
:::{admonition} Spacetime Notation: A Gentle Introduction
:class: dropdown note

The key idea is simple: **space and time are two aspects of the same thing**, and the index $\mu = 0, 1, 2, 3$ lets us treat them on equal footing.

**Four-vectors.** Each physical quantity has a time component ($\mu = 0$) and three spatial components ($\mu = 1, 2, 3$):

| Four-vector | $\mu = 0$ (time) | $\mu = 1,2,3$ (space) |
|:---|:---|:---|
| Coordinates $x^\mu$ | $x^0 = t$ | $\boldsymbol{x} = (x^1, x^2, x^3)$ |
| Derivatives $\partial_\mu$ | $\partial_0 = \partial_t$ | $\nabla = (\partial_1, \partial_2, \partial_3)$ |
| Energy-momentum $p^\mu$ | $p^0 = E$ | $\boldsymbol{p} = (p^1, p^2, p^3)$ |
| Gauge field $A_\mu$ | $A_0 = \Phi$ | $\boldsymbol{A} = (A_1, A_2, A_3)$ |
| Covariant derivative $D_\mu$ | $D_0 = \partial_t + \mathrm{i}(q/\hbar)\Phi$ | $\boldsymbol{D} = \nabla - \mathrm{i}(q/\hbar)\boldsymbol{A}$ |

Notice the pattern: every familiar pair (energy and momentum, scalar and vector potential, time and space derivatives) is unified into a single four-vector.

**Index conventions.** Upper and lower indices are related by the Lorentz metric $g_{\mu\nu} = \mathrm{diag}(-1, +1, +1, +1)$:

$$
a_\mu = g_{\mu\nu}\, a^\nu \quad \Longrightarrow \quad a_0 = -a^0, \quad a_i = a^i \; (i = 1,2,3)
$$

Repeated indices (one upper, one lower) are summed over: $a^\mu b_\mu = -a^0 b^0 + a^1 b^1 + a^2 b^2 + a^3 b^3$. This is the **Einstein summation convention**.

**The field strength tensor.** With this notation, $F_{\mu\nu} = \partial_\mu A_\nu - \partial_\nu A_\mu$ is a $4 \times 4$ antisymmetric matrix whose components are exactly $\boldsymbol{E}$ and $\boldsymbol{B}$:

$$
F_{\mu\nu} = \begin{pmatrix} 0 & E_x & E_y & E_z \\ -E_x & 0 & -B_z & B_y \\ -E_y & B_z & 0 & -B_x \\ -E_z & -B_y & B_x & 0 \end{pmatrix}
$$

For example, $F_{01} = \partial_0 A_1 - \partial_1 A_0 = \partial_t A_x + \partial_x \Phi = -(-\partial_x \Phi - \partial_t A_x) = E_x$, and $F_{12} = \partial_1 A_2 - \partial_2 A_1 = -(\nabla \times \boldsymbol{A})_z = -B_z$.

The beauty of $F_{\mu\nu}$ is that it packages six quantities ($\boldsymbol{E}$ and $\boldsymbol{B}$) into one geometric object that transforms simply under changes of reference frame.
:::

### The Gauge Principle Requires Electromagnetism

:::{admonition} The Gauge Principle
:class: important

Starting from only two inputs:

1. Local U(1) phase invariance: $\psi(\boldsymbol{x},t) \to \mathrm{e}^{\mathrm{i}q\chi(\boldsymbol{x},t)}\psi(\boldsymbol{x},t)$
2. The Schrödinger equation retains its form

we were **forced** to introduce a gauge field $\boldsymbol{A}$, replace $\nabla \to \nabla - \mathrm{i}(q/\hbar)\boldsymbol{A}$ (covariant derivative), and adopt $\hat{H} = \frac{(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2}{2m} + q\Phi$. From this, the Lorentz force and the electric/magnetic fields emerged automatically.

**Symmetry $\to$ Interaction**: The gauge field is not freely chosen—it is required by quantum consistency. Every charged particle couples to the same universal gauge field.
:::

### Summary

- **Kinetic momentum** $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ determines velocity; its components do not commute in a magnetic field.
- The **Lorentz force** $m\boldsymbol{a} = q(\boldsymbol{E} + \boldsymbol{v}\times\boldsymbol{B})$ is derived from the gauge-invariant Hamiltonian, not assumed.
- **Electric and magnetic fields** arise from gauge potentials: $\boldsymbol{E} = -\nabla\Phi - \partial_t\boldsymbol{A}$, $\boldsymbol{B} = \nabla\times\boldsymbol{A}$.
- The **field strength tensor** $F_{\mu\nu} = \partial_\mu A_\nu - \partial_\nu A_\mu$ is gauge-invariant.
- The **gauge principle** transforms symmetry into interaction: local U(1) invariance requires electromagnetism.
