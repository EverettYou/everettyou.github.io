## Lecture Notes

### Overview

Global phase is unobservable ([§1.1.1](../ch1_qubit/1-1-1-what-is-a-qubit)). What happens when we promote the phase to a function of position? The Schrödinger equation breaks—and fixing it forces us to introduce a gauge field and a new kind of derivative.

### From Global to Local

Under a **global** phase rotation $\psi \to \mathrm{e}^{\mathrm{i}\alpha}\psi$ ($\alpha$ constant), all observables are unchanged—this is the U(1) global symmetry ([§1.1.1](../ch1_qubit/1-1-1-what-is-a-qubit)). If physics is truly local, the description should not depend on a choice of phase convention at each point. This is the **local phase ambiguity**.

Promote $\alpha$ to a function of position:

$$
\psi(\boldsymbol{r}) \to \psi'(\boldsymbol{r}) = \mathrm{e}^{\mathrm{i}\alpha(\boldsymbol{r})}\psi(\boldsymbol{r})
$$

The probability density is still invariant:

$$
|\psi'(\boldsymbol{r})|^2 = |\mathrm{e}^{\mathrm{i}\alpha(\boldsymbol{r})}\psi(\boldsymbol{r})|^2 = |\psi(\boldsymbol{r})|^2
$$

But check the Schrödinger equation $\hat{H}\psi = E\psi$ with $\hat{H} = -\frac{\hbar^2}{2m}\nabla^2$. We say an equation is **covariant** under a transformation if it keeps the same form. For a global phase, $\mathrm{e}^{\mathrm{i}\alpha}$ factors through $\nabla^2$ and cancels—covariance holds. For a local phase, the derivative picks up an extra term:

$$
\partial_i\psi' = \mathrm{e}^{\mathrm{i}\alpha}\left(\partial_i + \mathrm{i}\,\partial_i\alpha\right)\psi
$$

The gradient of $\alpha$ does not cancel. The Laplacian is worse:

$$
\nabla^2\psi' = \mathrm{e}^{\mathrm{i}\alpha}\left[\nabla^2 + 2\mathrm{i}(\nabla\alpha) \cdot \nabla + \mathrm{i}(\nabla^2\alpha) - (\nabla\alpha) \cdot (\nabla\alpha)\right]\psi
$$

:::{admonition} The Gauge Problem
:class: important

Under $\psi \to \mathrm{e}^{\mathrm{i}\alpha(\boldsymbol{r})}\psi$, the Schrödinger equation is **not covariant**:

$$
-\frac{\hbar^2}{2m}\nabla^2\psi' \neq \mathrm{e}^{\mathrm{i}\alpha}\left(-\frac{\hbar^2}{2m}\nabla^2\psi\right)
$$

Global phase invariance does not extend to local phase invariance. If we want local phase to be as unobservable as global phase, the theory must be modified.
:::

### The Covariant Derivative

The extra terms come from $\partial_i$ acting on $\mathrm{e}^{\mathrm{i}\alpha}$. The fix: introduce a vector field $\boldsymbol{A}(\boldsymbol{r})$—the **gauge field**—and define the **covariant derivative**

$$
D_i = \partial_i - \mathrm{i}\frac{q}{\hbar}A_i
$$

where $q$ is the particle's charge. If we require $\boldsymbol{A}$ to shift as

$$
\boldsymbol{A} \to \boldsymbol{A} + \frac{\hbar}{q}\nabla\alpha
$$

whenever we rotate $\psi \to \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi$, then the shift in $\boldsymbol{A}$ cancels the unwanted gradient. This compensating shift is called a **gauge transformation**.

:::{admonition} Covariant Derivative
:class: important

Under $\psi \to \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi$ and $\boldsymbol{A} \to \boldsymbol{A} + \nabla\alpha$:

$$
D_i\psi \to \mathrm{e}^{\mathrm{i}q\alpha/\hbar}(D_i\psi)
$$

The covariant derivative transforms like $\psi$ itself—unlike the ordinary derivative $\partial_i$.
:::

:::{admonition} Derivation: Covariance of $D_i$
:class: dropdown information

Under $\psi' = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi$ and $\boldsymbol{A}' = \boldsymbol{A} + \nabla\alpha$:

$$
\begin{split}
D_i'\psi' &= \left(\partial_i - \mathrm{i}\frac{q}{\hbar}(A_i + \partial_i\alpha)\right)\mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi \\
&= \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\left(\mathrm{i}\frac{q}{\hbar}\partial_i\alpha\right)\psi + \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\partial_i\psi - \mathrm{i}\frac{q}{\hbar}A_i\,\mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi - \mathrm{i}\frac{q}{\hbar}(\partial_i\alpha)\mathrm{e}^{\mathrm{i}q\alpha/\hbar}\psi
\end{split}
$$

The first and last terms cancel:

$$
D_i'\psi' = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}\left(\partial_i - \mathrm{i}\frac{q}{\hbar}A_i\right)\psi = \mathrm{e}^{\mathrm{i}q\alpha/\hbar}(D_i\psi)
$$

:::

### Gauge-Invariant Hamiltonian

Replace every $\partial_i$ with $D_i$—the **minimal coupling prescription**:

$$
\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla \;\longrightarrow\; \hat{\boldsymbol{p}} - q\boldsymbol{A} = -\mathrm{i}\hbar\boldsymbol{D}
$$

$$
\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + V(\boldsymbol{r})
$$

Since $D_i\psi$ transforms like $\psi$, so does $D_iD_j\psi$, and the Schrödinger equation with this Hamiltonian is covariant under local gauge transformations.

:::{admonition} The Gauge Principle
:class: important

Requiring local phase covariance **forces**:

1. A gauge field $\boldsymbol{A}(\boldsymbol{r})$ with $\boldsymbol{A} \to \boldsymbol{A} + \nabla\alpha$.
2. The covariant derivative $D_i = \partial_i - \mathrm{i}(q/\hbar)A_i$.
3. The kinetic energy $\frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2$.

The coupling to $\boldsymbol{A}$ is a consequence of the symmetry, not an assumption.
:::

### Kinetic Momentum

The combination $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A}$ that appears in the Hamiltonian has a direct physical meaning. From Hamilton's equation $\dot{\boldsymbol{r}} = \partial H / \partial \boldsymbol{p}$:

$$
m\dot{\boldsymbol{r}} = \hat{\boldsymbol{p}} - q\boldsymbol{A} = \hat{\boldsymbol{\pi}}
$$

so $\hat{\boldsymbol{\pi}}$ is the **kinetic momentum**—it equals $m\boldsymbol{v}$, the mass times velocity. The canonical momentum $\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla$ includes an additional contribution $q\boldsymbol{A}$ from the gauge field. The distinction between canonical and kinetic momentum, and which quantities are gauge-invariant, will be discussed in [§4.1.3](4-1-3-gauge-invariance).

:::{admonition} Expansion of the Kinetic Energy
:class: dropdown example

**Problem.** Expand $(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2$ and identify the couplings.

**Solution.**

$$
(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 = \hat{\boldsymbol{p}}^2 - q(\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}}) + q^2\boldsymbol{A}^2
$$

Using $\hat{\boldsymbol{p}} = -\mathrm{i}\hbar\nabla$:

$$
\hat{\boldsymbol{p}} \cdot \boldsymbol{A} + \boldsymbol{A} \cdot \hat{\boldsymbol{p}} = -\mathrm{i}\hbar(\nabla \cdot \boldsymbol{A} + 2\boldsymbol{A} \cdot \nabla)
$$

In the Coulomb gauge $\nabla \cdot \boldsymbol{A} = 0$:

$$
\frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 = \frac{\hat{\boldsymbol{p}}^2}{2m} - \frac{q}{m}\boldsymbol{A} \cdot \hat{\boldsymbol{p}} + \frac{q^2}{2m}\boldsymbol{A}^2
$$

The linear term is the **paramagnetic** coupling; the quadratic term is the **diamagnetic** coupling.
:::

:::{admonition} Poll: Local phase and the ordinary derivative
:class: dropdown poll

Under $\psi(\boldsymbol{r}) \to \mathrm{e}^{\mathrm{i}\alpha(\boldsymbol{r})}\psi(\boldsymbol{r})$, the ordinary derivative transforms as $\partial_i\psi' = \mathrm{e}^{\mathrm{i}\alpha}(\partial_i + \mathrm{i}\partial_i\alpha)\psi$. What is the role of $A_i$ in the covariant derivative $D_i = \partial_i - \mathrm{i}(q/\hbar)A_i$?

(A) $A_i$ shifts as $A_i \to A_i + (\hbar/q)\partial_i\alpha$, cancelling the extra $\mathrm{i}(\partial_i\alpha)\psi$ term so that $D_i\psi$ transforms like $\psi$.

(B) $A_i$ makes the wavefunction real-valued, eliminating the need for complex phases.

(C) $A_i$ ensures that $\partial_i\psi' = \mathrm{e}^{\mathrm{i}\alpha}\partial_i\psi$ without any additional terms.

(D) $A_i$ is a mathematical convenience with no physical consequence—it can always be set to zero.

:::

### Summary

- **Local phase ambiguity** (motivated by local frame independence) breaks the eigenvalue equation because $\partial_i$ acting on $\mathrm{e}^{\mathrm{i}\alpha}$ generates extra terms.
- The **covariant derivative** $D_i = \partial_i - \mathrm{i}(q/\hbar)A_i$, with $\boldsymbol{A} \to \boldsymbol{A} + \nabla\alpha$, transforms like $\psi$ and restores covariance.
- The **gauge principle**: local phase covariance forces $\hat{H} = \frac{1}{2m}(\hat{\boldsymbol{p}} - q\boldsymbol{A})^2 + V$.
- The **kinetic momentum** $\hat{\boldsymbol{\pi}} = \hat{\boldsymbol{p}} - q\boldsymbol{A} = m\boldsymbol{v}$ is the physical momentum that determines velocity.
