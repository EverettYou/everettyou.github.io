## Lecture Notes

Huygens' principle reveals that light propagation is fundamentally about superposition of wavelets, each carrying a phase proportional to optical path length. In the high-frequency limit, constructive interference from paths of stationary optical path length produces Fermat's principle—the geometric optics emerges from wave physics.

### Huygens' Principle: The Foundation of Wave Optics

**Christiaan Huygens** (1690) introduced a constructive principle for understanding wave propagation:

:::{admonition} Huygens' Principle
:class: important

Each point on a wavefront acts as a source of **secondary wavelets**. These wavelets are spherical waves emanating from that point. The wavefront at a later time is the **envelope** (coherent superposition) of all these secondary wavelets.
:::

This principle is not derived from Maxwell's equations (which didn't exist in Huygens' time) but rather a geometric statement about how waves evolve. Yet it captures the essence of wave behavior.

#### Elementary Wavelet and Phase

Consider a point source on the wavefront at position $\boldsymbol{r}_0$ at time $t=0$. It emits a spherical wave that reaches position $\boldsymbol{r}$ at time $t$:

$$

\psi(\boldsymbol{r}, t) \propto \frac{e^{i(k|\boldsymbol{r}-\boldsymbol{r}_0| - \omega t)}}{|\boldsymbol{r}-\boldsymbol{r}_0|}

$$ (eq-spherical-wave)

The **phase** of this wavelet is

$$

\Phi(\boldsymbol{r}, t) = k|\boldsymbol{r}-\boldsymbol{r}_0| - \omega t = \frac{2\pi}{\lambda}|\boldsymbol{r}-\boldsymbol{r}_0| - 2\pi f t

$$ (eq-wavelet-phase)

where $\lambda = 2\pi/k$ is the wavelength and $f = \omega/(2\pi)$ is the frequency. The distance traveled is $|\boldsymbol{r}-\boldsymbol{r}_0|$, and the phase accumulated is proportional to this distance (in units of wavelength).

#### In an Inhomogeneous Medium

When light propagates through a medium with varying refractive index $n(\boldsymbol{r})$, the **optical path length** from $\boldsymbol{r}_0$ to $\boldsymbol{r}$ is

$$

L(\boldsymbol{r}_0 \to \boldsymbol{r}) = \int_{\boldsymbol{r}_0}^{\boldsymbol{r}} n(\boldsymbol{r}') \, ds'

$$ (eq-optical-path-length)

where $ds'$ is the arc length along the path. The phase accumulated is

$$

\Phi = k L(\boldsymbol{r}_0 \to \boldsymbol{r}) - \omega t = \frac{2\pi}{\lambda_0} L(\boldsymbol{r}_0 \to \boldsymbol{r}) - 2\pi f t

$$ (eq-optical-path-phase)

where $\lambda_0$ is the **vacuum wavelength**. This is crucial: **Huygens' principle weights each secondary wavelet by its phase, which depends on optical path length**.

### Deriving Refraction from Huygens' Principle

Consider a plane wavefront in medium 1 (refractive index $n_1$) incident on a boundary with medium 2 (refractive index $n_2$). Different points on the wavefront reach the boundary at different times. Each point becomes a secondary source.

**Setup**: 
- Point $A$ on the wavefront reaches the boundary at time $t=0$.
- Point $B$ on the same wavefront, a distance $\ell$ away, reaches the boundary at time $t_B = \ell/(v_1) = \ell n_1/c$ (where $v_1 = c/n_1$).

**Secondary wavelets**:
- At point $A$: wavelet propagates a distance $d_A$ into medium 2 in time $t_B$, so $d_A = (c/n_2) t_B = \ell n_1/n_2$.
- At point $B$: wavelet has just started propagating, so $d_B = 0$.

**New wavefront**: The envelope of secondary wavelets is a plane at angle $\theta_2$ to the normal, where

$$

\sin\theta_2 = \frac{d_A}{\ell} = \frac{n_1}{n_2} \sin\theta_1

$$ (eq-snell-law)

This is **Snell's law of refraction**, derived purely from geometry and Huygens' principle.

### Interference and Diffraction

Huygens' principle naturally explains interference and diffraction because it is fundamentally about **superposition of waves**.

#### Young's Double-Slit Experiment

Two slits act as coherent secondary sources (from a common incident wavefront). Each slit emits spherical wavelets. At a point $P$ on a distant screen:

**Path difference**: The distance from slit 1 to $P$ is $r_1$; from slit 2 is $r_2$. The path difference is $\Delta r = r_2 - r_1$.

**Phase difference**: The phase accumulated from each wavelet is $k\Delta r$. 

**Interference condition**:

- **Constructive**: $\Delta r = m\lambda$ (phase difference $2\pi m$), $m = 0, \pm 1, \pm 2, \ldots$ → bright fringe.

- **Destructive**: $\Delta r = (m + 1/2)\lambda$ (phase difference $(2m+1)\pi$) → dark fringe.

#### Fresnel Diffraction from an Edge

When light passes an edge, secondary wavelets from all unobstructed points on the wavefront contribute to a point beyond the geometric shadow. The **Fresnel zones** are regions where the phase difference from neighboring zones is $\pi$ (path difference $\lambda/2$). Summing contributions from Fresnel zones shows that:

1. The intensity near the shadow boundary oscillates (Fresnel fringes).
2. Inside the geometric shadow, there is non-zero intensity from contributions outside the shadow.
3. This is purely a **phase interference effect**—the geometric optics picture (straight-line rays) breaks down.

:::{admonition} Poisson Spot (1826)
:class: example

When Fresnel's diffraction theory predicted that light would appear at the **center of the shadow of a circular disk**, it seemed absurd. The physicist Poisson used this as an argument against wave theory. But experimentalist François Arago verified the prediction: there is indeed a bright spot at the center of the disk's shadow. This "Poisson spot" was crucial evidence for the wave nature of light and demonstrated the power of Huygens' principle.
:::

### The Eikonal Equation: Connecting Waves to Rays

Maxwell's equations describe electromagnetic waves in a medium with varying refractive index $n(\boldsymbol{r})$. For high-frequency waves (short wavelength relative to variations in $n$), we can use the **geometric optics approximation**.

**Wave equation**:

$$

\nabla^2 \boldsymbol{E} = \frac{n^2}{c^2} \frac{\partial^2 \boldsymbol{E}}{\partial t^2}

$$ (eq-wave-equation)

**WKB ansatz**: Write the field as

$$

\boldsymbol{E}(\boldsymbol{r}, t) = \boldsymbol{E}_0(\boldsymbol{r}) \, e^{i(\psi(\boldsymbol{r}) - \omega t)}

$$ (eq-eikonal-ansatz)

where $\psi(\boldsymbol{r})$ is the **eikonal function** (related to optical path length) and $\boldsymbol{E}_0$ is a slowly-varying envelope.

Substituting into Maxwell's equations and keeping only the leading-order (high-frequency) terms, we obtain the **eikonal equation**:

$$

(\nabla\psi)^2 = n(\boldsymbol{r})^2

$$ (eq-eikonal-equation)

**Light rays** are the orthogonal trajectories to surfaces of constant eikonal, i.e., $\nabla \psi$ points along the ray direction.

:::{admonition} Ray trajectories from the eikonal
:class: note

A light ray follows the direction of increasing eikonal:

$$

\frac{d\boldsymbol{r}}{ds} = \frac{\nabla\psi}{|\nabla\psi|} = \frac{\nabla\psi}{n}

$$

where $s$ is arc length along the ray. Differentiating the eikonal equation and using the above, we recover **Newton's law for light rays** in an inhomogeneous medium:

$$

\frac{d}{ds}\left(n \frac{d\boldsymbol{r}}{ds}\right) = \nabla n

$$ (eq-newtons-law-light)

This shows that light is **bent by the gradient of the refractive index**, just as particles are bent by forces.
:::

### From Huygens to Fermat: The Stationary Phase Connection

Huygens' principle appears to sum contributions from all possible paths (secondary wavelets from all points on the wavefront). Fermat's principle selects only the extremal path. How are they related?

**Huygens' sum**: Consider a wavefront propagating from $A$ to $B$ through an inhomogeneous medium. Each point on an intermediate wavefront can be viewed as a secondary source. The amplitude at $B$ is the superposition:

$$

\psi_B \propto \int \exp(i k L(A \to \text{intermediate point} \to B)) \, d(\text{intermediate point})

$$ (eq-huygens-sum)

where $L$ is the optical path length. This is essentially an **integral over all possible paths**, weighted by phase $e^{iL}$.

**Stationary phase argument**: When the frequency is high (short wavelength, large $k$), the phase $kL$ oscillates rapidly for paths far from an extremum. Only paths where $\delta L = 0$ (Fermat's principle) have slowly-varying phase and contribute constructively to the integral. All other paths cancel through destructive interference.

**Therefore: Fermat's principle emerges from Huygens' principle in the high-frequency (short-wavelength) limit.** Geometric optics is the **classical limit of wave optics**.

| Optics regime | Principle | Limit |
|--------------|-----------|-------|
| Wave (low frequency) | Huygens' superposition of all wavelets | All paths contribute |
| Transition (intermediate) | Stationary phase approximation | Paths near extremum dominate |
| Ray (high frequency) | Fermat's extremal path | Single classical path |

### Quantitative Connection: Huygens Integral

The rigorous connection is the **Huygens-Fresnel integral**. For light propagating from a source at $A$ to an observation point $B$, the amplitude is

$$

\psi_B = \frac{1}{i\lambda} \int_S K(\boldsymbol{r}) \, e^{ikL(A \to \boldsymbol{r} \to B)} \, dS

$$ (eq-huygens-fresnel)

where:
- The integral is over an aperture surface $S$ (the unobstructed part of the intermediate wavefront).
- $L$ is the optical path length $L = \int n \, ds$.
- $K(\boldsymbol{r})$ is an obliquity factor (slowly-varying).

**Key observation**: The phase is $e^{ikL} = e^{i L/\lambda_0}$ where $\lambda_0 = 2\pi/k$ is the wavelength. This is a **sum over paths weighted by their phase**—the precursor to Feynman's path integral in quantum mechanics.

### Summary

- **Huygens' principle**: Secondary wavelets emanate from all points on a wavefront; their superposition gives the next wavefront.

- **Phase and optical path**: Each wavelet carries phase proportional to optical path length $L = \int n \, ds$.

- **Interference and diffraction**: Superposition of wavelets naturally produces interference fringes and diffraction patterns.

- **Eikonal equation**: Governs ray trajectories; reduces to Newton's law for light in non-uniform media.

- **Stationary phase limit**: In the high-frequency limit, only paths with extremal optical path length survive—Fermat's principle emerges from Huygens' principle.

- **Path integral precursor**: The Huygens-Fresnel integral is a sum over paths with phase weights—foundational to understanding Feynman's path integral in quantum mechanics.


:::{admonition} Derivation: Snell's Law from Fermat's Principle via Calculus of Variations
:class: dropdown note

Fermat's principle states that light follows the path for which the optical path length is stationary (extremal). Given a path from point $A$ in medium 1 (refractive index $n_1$) to point $B$ in medium 2 (refractive index $n_2$), crossing the interface at position $x$, the optical path length is

$$L(x) = n_1\sqrt{d_1^2 + x^2} + n_2\sqrt{d_2^2 + (w-x)^2}

$$where $d_1$, $d_2$ are the perpendicular distances from $A$ and $B$ to the interface, and $w$ is the separation between the perpendiculars.

By Fermat's principle, at the extremum:

$$rac{dL}{dx} = 0

$$Computing the derivative:

$$rac{dL}{dx} = n_1 rac{x}{\sqrt{d_1^2 + x^2}} - n_2 rac{w-x}{\sqrt{d_2^2 + (w-x)^2}}

$$At the extremum, this vanishes. Note that $\sin	heta_1 = x/\sqrt{d_1^2 + x^2}$ and $\sin	heta_2 = (w-x)/\sqrt{d_2^2 + (w-x)^2}$ are the angles of incidence and refraction. Therefore:

$$n_1 \sin	heta_1 = n_2 \sin	heta_2

$$

This is **Snell's law**, derived directly from Fermat's principle using calculus of variations. The statement $\delta L = 0$ leads immediately to the law of refraction.

:::
