## Lecture Notes

### Overview

The Schrödinger equation is traditionally *postulated* as a fundamental axiom. Here we *derive* it from the path integral by taking the limit $\delta t \to 0$ in the time-slice propagator formula. This reveals that the two formulations of quantum mechanics—the path integral (global sum over histories) and the Schrödinger equation (local differential equation)—are mathematically equivalent. The propagator emerges as the Green's function of the Schrödinger equation.

### The Time-Slice Propagator Revisited

Recall from §3.2.1 that for a small time interval $\delta t$, the propagator factorizes as

$$
\psi(x, t + \delta t) = \int K_{\delta t}^{\text{free}}(x, x') \, \mathrm{e}^{-\mathrm{i}V(x') \delta t / \hbar} \, \psi(x', t) \, \mathrm{d}x'
$$

where the free-particle (kinetic) kernel is

$$
K_{\delta t}^{\text{free}}(x, x') = \left(\frac{m}{2\pi \mathrm{i}\hbar \delta t}\right)^{1/2} \exp\left(\frac{\mathrm{i}m(x - x')^2}{2\hbar \delta t}\right)
$$ (eq-K-free-delta-t-one-step)

To derive the Schrödinger equation, we expand both sides in powers of $\delta t$ and identify the $O(\delta t)$ term.

### Step 1: Gaussian Integral and Taylor Expansion

Substitute $\delta x = x - x'$. The kinetic phase

$$
\exp\left(\frac{\mathrm{i} m(\delta x)^2}{2\hbar \delta t}\right)
$$

oscillates rapidly unless $|\delta x| \sim \sqrt{\hbar \delta t / m}$, a width that shrinks to zero as $\delta t \to 0$. This allows us to Taylor-expand $\psi(x', t)$ around $x'$:

$$
\psi(x' + \delta x, t) = \psi(x', t) + \delta x \cdot \frac{\partial}{\partial x} \psi(x', t) + \frac{1}{2}(\delta x)^2 \frac{\partial^2 \psi}{\partial x^2} + \ldots
$$

Changing variables to $x' = x - \delta x$:

$$
\psi(x, t + \delta t) = \int \mathrm{d}(\delta x) \, K_{\delta t}^{\text{free}}(\delta x) \left[ \psi + \delta x \cdot \frac{\partial}{\partial x} \psi + \frac{1}{2}(\delta x)^2 \partial_x^2 \psi \right] \mathrm{e}^{-\mathrm{i}V(x) \delta t / \hbar}
$$

where $K_{\delta t}^{\text{free}}(\delta x) = (m / 2\pi \mathrm{i}\hbar \delta t)^{1/2} \exp(\mathrm{i}m(\delta x)^2 / 2\hbar \delta t)$.

### Step 2: Gaussian Moments

:::{admonition} Key Gaussian Integrals
:class: note

For the normalized Gaussian kernel $K_{\delta t}^{\text{free}}(\delta x) = \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2} \exp\!\left(\frac{\mathrm{i}m(\delta x)^2}{2\hbar\,\delta t}\right)$:

$$
\int K_{\delta t}^{\text{free}}(\delta x) \, \mathrm{d}(\delta x) = 1
$$

$$
\int K_{\delta t}^{\text{free}}(\delta x) \, \delta x \, \mathrm{d}(\delta x) = 0 \quad \text{(odd integrand)}
$$

$$
\int K_{\delta t}^{\text{free}}(\delta x) \, (\delta x)^2 \, \mathrm{d}(\delta x) = \frac{\mathrm{i}\hbar\,\delta t}{m}
$$

:::

:::{admonition} Derivation: Gaussian Moments
:class: dropdown information

**Step 1: The fundamental Gaussian integral (from scratch).**
Consider $I = \int_{-\infty}^{\infty} \mathrm{e}^{-u^2}\,\mathrm{d}u$. The trick is to compute $I^2$ by switching to polar coordinates:

$$
I^2 = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} \mathrm{e}^{-(u^2+v^2)}\,\mathrm{d}u\,\mathrm{d}v = \int_0^{2\pi}\int_0^{\infty} \mathrm{e}^{-r^2}\,r\,\mathrm{d}r\,\mathrm{d}\theta = 2\pi \cdot \frac{1}{2} = \pi
$$

where $\int_0^\infty r\,\mathrm{e}^{-r^2}\,\mathrm{d}r = \frac{1}{2}$ follows from the substitution $s = r^2$. Therefore $I = \sqrt{\pi}$.

**Step 2: Introduce a coefficient $\alpha$.**
Substitute $u = \sqrt{-\alpha}\,x$ (valid for $\operatorname{Re}(-\alpha) > 0$, extended to complex $\alpha$ by analytic continuation):

$$
\int_{-\infty}^{\infty} \mathrm{e}^{\alpha x^2}\,\mathrm{d}x = \frac{1}{\sqrt{-\alpha}}\int_{-\infty}^{\infty} \mathrm{e}^{-u^2}\,\mathrm{d}u = \sqrt{\frac{\pi}{-\alpha}}
$$

**Step 3: Generate higher moments by differentiating in $\alpha$.**
Differentiate both sides with respect to $\alpha$:

$$
\frac{\mathrm{d}}{\mathrm{d}\alpha}\int \mathrm{e}^{\alpha x^2}\,\mathrm{d}x = \int x^2\,\mathrm{e}^{\alpha x^2}\,\mathrm{d}x = \frac{\mathrm{d}}{\mathrm{d}\alpha}\sqrt{\frac{\pi}{-\alpha}} = \frac{1}{2}(-\alpha)^{-3/2}\sqrt{\pi}
$$

Repeating: $\int x^{2n}\mathrm{e}^{\alpha x^2}\mathrm{d}x = \frac{(2n-1)!!}{2^n}(-\alpha)^{-(2n+1)/2}\sqrt{\pi}$ (all odd moments vanish by symmetry).

**Step 4: Apply to the short-time propagator.**
Write $\alpha = \mathrm{i}m/(2\hbar\,\delta t)$ so that $K_{\delta t}^{\text{free}}(\delta x) = (\alpha/\pi)^{1/2}\,\mathrm{e}^{\alpha(\delta x)^2}$.

- *Zeroth moment*: $\int K_{\delta t}^{\text{free}}\,\mathrm{d}(\delta x) = (\alpha/\pi)^{1/2}\sqrt{\pi/(-\alpha)} = 1$ (normalization).
- *First moment*: vanishes by symmetry (odd integrand).
- *Second moment*:

$$
\int K_{\delta t}^{\text{free}}(\delta x)\,(\delta x)^2\,\mathrm{d}(\delta x) = (\alpha/\pi)^{1/2} \cdot \frac{\sqrt{\pi}}{2(-\alpha)^{3/2}} = \frac{1}{-2\alpha} = \frac{\mathrm{i}\hbar\,\delta t}{m}
$$

This is $O(\delta t)$, which is why it contributes at leading order in the Taylor expansion of $\psi(x + \delta x)$ that produces the Schrödinger equation.

- *Fourth moment*: $\int (\delta x)^4\,K_{\delta t}^{\text{free}}\,\mathrm{d}(\delta x) = \frac{3}{4\alpha^2} = O(\delta t^2)$, which is subleading and can be dropped.

:::

### Step 3: Collecting O(δt) Terms

Using the moments from Step 2:

- The zeroth-order term: $\int K_{\delta t}^{\text{free}} \psi \, \mathrm{e}^{-\mathrm{i}V \delta t/\hbar} \approx \psi(1 - \mathrm{i}V \delta t/\hbar) = \psi - (\mathrm{i}V \delta t/\hbar) \psi$
- The first-moment term vanishes: $\int K_{\delta t}^{\text{free}} \, \delta x \cdot \frac{\partial}{\partial x} \psi \, \ldots = 0$
- The second-moment term: $\int K_{\delta t}^{\text{free}} \, (\delta x)^2 \, \partial_i \partial_j \psi \, \ldots = -\frac{\mathrm{i}\hbar \delta t}{m} \frac{\partial^2}{\partial x^2} \psi$
:::{admonition} Derivation: Collecting the O(δt) Terms
:class: dropdown information

Starting from the integral after substitution $\delta x = x - x'$:

$$
\psi(x, t + \delta t) = \mathrm{e}^{-\mathrm{i}V(x)\,\delta t/\hbar}\!\int \mathrm{d}(\delta x)\,K_{\delta t}^{\text{free}}(\delta x)\!\left[\psi + \delta x\,\partial_x\psi + \tfrac{1}{2}(\delta x)^2\,\partial_x^2\psi + \cdots\right]
$$

Evaluate each term using the Gaussian moments:

**Zeroth-order term.** The $\psi$ piece gives

$$
\mathrm{e}^{-\mathrm{i}V\delta t/\hbar}\!\int K_{\delta t}^{\text{free}}(\delta x)\,\psi\,\mathrm{d}(\delta x) = \mathrm{e}^{-\mathrm{i}V\delta t/\hbar}\,\psi
$$

since the kernel is normalized to unity. Expanding the exponential to $O(\delta t)$:

$$
\mathrm{e}^{-\mathrm{i}V\delta t/\hbar}\,\psi = \psi - \frac{\mathrm{i}}{\hbar}V\,\delta t\,\psi + O((\delta t)^2)
$$

**First-order term.** The $\delta x\,\partial_x\psi$ piece vanishes because the first moment of the Gaussian kernel is zero (odd integrand):

$$
\mathrm{e}^{-\mathrm{i}V\delta t/\hbar}\!\int K_{\delta t}^{\text{free}}(\delta x)\,\delta x\,\partial_x\psi\,\mathrm{d}(\delta x) = 0
$$

**Second-order term.** The $\tfrac{1}{2}(\delta x)^2\,\partial_x^2\psi$ piece contributes at $O(\delta t)$ via the second Gaussian moment. To leading order we drop the $\mathrm{e}^{-\mathrm{i}V\delta t/\hbar}$ correction (which would produce $O((\delta t)^2)$):

$$
\tfrac{1}{2}\!\int K_{\delta t}^{\text{free}}(\delta x)\,(\delta x)^2\,\partial_x^2\psi\,\mathrm{d}(\delta x) = \frac{1}{2}\cdot\frac{\mathrm{i}\hbar\,\delta t}{m}\,\partial_x^2\psi
$$

**Combining.** Collecting all $O(\delta t)$ contributions:

$$
\psi(x, t+\delta t) = \psi + \frac{\mathrm{i}\hbar\,\delta t}{2m}\,\partial_x^2\psi - \frac{\mathrm{i}}{\hbar}V\,\delta t\,\psi + O((\delta t)^2)
$$

Moving $\psi$ to the left and dividing by $\delta t$:

$$
\frac{\psi(x, t+\delta t) - \psi(x,t)}{\delta t} = \frac{\mathrm{i}\hbar}{2m}\frac{\partial^2\psi}{\partial x^2} - \frac{\mathrm{i}}{\hbar}V(x)\,\psi
$$

Taking $\delta t \to 0$ on the left gives $\partial\psi/\partial t$. Multiplying both sides by $\mathrm{i}\hbar$:

$$
\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = -\frac{\hbar^2}{2m}\frac{\partial^2\psi}{\partial x^2} + V(x)\,\psi = \hat{H}\psi
$$

which is the time-dependent Schrödinger equation.
:::


Therefore:

$$
\psi(x, t + \delta t) - \psi(x, t) = -\frac{\mathrm{i}\hbar \delta t}{m} \frac{\partial^2}{\partial x^2} \psi - \frac{\mathrm{i}}{\hbar} V(x) \delta t \, \psi + O((\delta t)^2)
$$

Dividing by $\delta t$ and taking $\delta t \to 0$:

$$
\frac{\partial \psi}{\partial t} = \frac{\mathrm{i}\hbar}{2m} \frac{\partial^2}{\partial x^2} \psi - \frac{\mathrm{i}}{\hbar} V(x) \psi
$$

Multiplying by $\mathrm{i}\hbar$:

:::{admonition} The Schrödinger Equation
:class: important

$$
\mathrm{i}\hbar \frac{\partial \psi}{\partial t} = \left(-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x)\right) \psi = \hat{H} \psi
$$

The Schrödinger equation is **derived** from the path integral, not postulated. The kinetic term produces the Laplacian $\frac{\partial^2}{\partial x^2}$, and the potential appears as a multiplicative factor.

:::

### Two Equivalent Formulations

The derivation reveals a deep equivalence:

:::{admonition} Path Integral ↔ Schrödinger Equation
:class: important

Quantum mechanics admits two mathematically equivalent formulations:

| Aspect | Path Integral | Schrödinger Equation |
|--------|---|---|
| **Perspective** | Global: sum over all histories | Local: differential equation |
| **Central Object** | Propagator $K = \sum \mathrm{e}^{\mathrm{i}S/\hbar}$ | Wavefunction $\psi(x, t)$ |
| **Equation Satisfied** | Schrödinger equation with delta-function initial condition | Schrödinger equation with arbitrary initial state |

The propagator $K(x, t_f; x', t_i)$ is the Green's function of the Schrödinger equation:

$$
\mathrm{i}\hbar \frac{\partial K}{\partial t_f} = \hat{H}_{x_f} K(x_f, t_f; x', t_i)
$$

with the initial condition $K(x_f, t_i; x', t_i) = \delta(x_f - x')$.
:::

:::{admonition} Derivation: Propagator Satisfies the Schrödinger Equation
:class: dropdown information

We verify that the propagator $K(x, t; x', t')$, defined by the path integral, satisfies the Schrödinger equation in its first argument with a delta-function initial condition.

**Step 1: Initial condition.** At $t = t'$, the time-slicing formula reduces to a single slice with $\delta t = 0$. The free-particle kernel becomes

$$
K_{\delta t}^{\text{free}}(x, x') = \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2}\exp\!\left(\frac{\mathrm{i}m(x-x')^2}{2\hbar\,\delta t}\right) \xrightarrow{\delta t\to 0} \delta(x - x')
$$

since a normalized Gaussian with variance $\to 0$ converges to a delta function. This gives the initial condition $K(x, t'; x', t') = \delta(x - x')$.

**Step 2: Time evolution.** For $t > t'$, the composition property of the propagator reads

$$
K(x, t+\delta t; x', t') = \int K_{\delta t}(x, x'')\,K(x'', t; x', t')\,\mathrm{d}x''
$$

This has exactly the same structure as $\psi(x, t+\delta t) = \int K_{\delta t}(x, x'')\,\psi(x'', t)\,\mathrm{d}x''$ with $\psi(x'',t) \equiv K(x'',t;x',t')$. The derivation of the Schrödinger equation from the time-slice formula therefore applies verbatim, giving

$$
\mathrm{i}\hbar\frac{\partial}{\partial t}K(x, t; x', t') = \hat{H}_x\,K(x, t; x', t')
$$

where $\hat{H}_x = -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x)$ acts on the first position argument.

**Step 3: Consistency.** The general solution to the Schrödinger equation is then recovered by superposition:

$$
\psi(x,t) = \int K(x,t;x',t')\,\psi(x',t')\,\mathrm{d}x'
$$

At $t = t'$, the delta-function initial condition ensures $\psi(x,t') = \int \delta(x-x')\,\psi(x',t')\,\mathrm{d}x' = \psi(x,t')$ — a self-consistency check.
:::

The wavefunction at any time is obtained by convolving the propagator with the initial state:

$$
\psi(x, t) = \int K(x, t; x', 0) \, \psi(x', 0) \, \mathrm{d}x'
$$

This shows that solving the Schrödinger equation is equivalent to evaluating the path integral.

:::{admonition} Derivation: Spectral Decomposition of the Propagator
:class: dropdown information

The propagator can be written explicitly in terms of the energy eigenstates. If $\hat{H}\psi_n(x) = E_n\psi_n(x)$, then

$$
K(x, t; x', 0) = \sum_n \psi_n(x)\,\psi_n^*(x')\,\mathrm{e}^{-\mathrm{i}E_n t/\hbar}
$$

**Step 1: Eigenfunction expansion.** Since $\{\psi_n\}$ forms a complete orthonormal basis, expand $K$ at fixed $(x', t)$ as

$$
K(x, t; x', 0) = \sum_n c_n(x', t)\,\psi_n(x)
$$

**Step 2: Time evolution from the Schrödinger equation.** Substitute into $\mathrm{i}\hbar\partial_t K = \hat{H}_x K$:

$$
\mathrm{i}\hbar\sum_n \dot{c}_n\,\psi_n = \sum_n c_n\,E_n\,\psi_n
$$

Projecting onto $\psi_m$ using orthonormality $\langle\psi_m|\psi_n\rangle = \delta_{mn}$ gives $\mathrm{i}\hbar\dot{c}_m = E_m c_m$, with solution

$$
c_n(x', t) = c_n(x', 0)\,\mathrm{e}^{-\mathrm{i}E_n t/\hbar}
$$

**Step 3: Initial condition.** At $t = 0$, the propagator is a delta function: $K(x, 0; x', 0) = \delta(x - x')$. So

$$
\sum_n c_n(x', 0)\,\psi_n(x) = \delta(x - x') = \sum_n \psi_n(x)\,\psi_n^*(x')
$$

where the second equality uses the completeness relation $\sum_n \psi_n(x)\psi_n^*(x') = \delta(x-x')$. Matching coefficients: $c_n(x', 0) = \psi_n^*(x')$.

**Step 4: Result.** Combining Steps 2 and 3:

$$
K(x, t; x', 0) = \sum_n \psi_n(x)\,\psi_n^*(x')\,\mathrm{e}^{-\mathrm{i}E_n t/\hbar}
$$

**Verification.** Substituting into the superposition formula:

$$
\psi(x, t) = \int K(x, t; x', 0)\,\psi(x', 0)\,\mathrm{d}x' = \sum_n \psi_n(x)\,\mathrm{e}^{-\mathrm{i}E_n t/\hbar}\underbrace{\int \psi_n^*(x')\,\psi(x', 0)\,\mathrm{d}x'}_{c_n(0)}
$$

This recovers the standard time-dependent superposition $\psi(x,t) = \sum_n c_n\,\psi_n(x)\,\mathrm{e}^{-\mathrm{i}E_n t/\hbar}$, confirming consistency.
:::


:::{admonition} Derivation: Probability Current and Local Probability Conservation
:class: dropdown information

The Schrödinger equation guarantees that probability is **locally conserved** — it cannot be created or destroyed, only transported. We derive the **continuity equation**

$$
\frac{\partial\rho}{\partial t} + \nabla\cdot\boldsymbol{j} = 0
$$

where $\rho = \vert\psi\vert^2$ is the probability density and $\boldsymbol{j}$ is the probability current.

**Step 1: Time derivative of the probability density.**

Write the Schrödinger equation and its complex conjugate:

$$
\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = \left(-\frac{\hbar^2}{2m}\nabla^2 + V\right)\psi, \qquad -\mathrm{i}\hbar\frac{\partial\psi^*}{\partial t} = \left(-\frac{\hbar^2}{2m}\nabla^2 + V\right)\psi^*
$$

Compute $\partial_t\rho = \partial_t(\psi^*\psi) = (\partial_t\psi^*)\psi + \psi^*(\partial_t\psi)$. Substituting the equations above:

$$
\frac{\partial\rho}{\partial t} = \frac{\mathrm{i}\hbar}{2m}\bigl(\psi\nabla^2\psi^* - \psi^*\nabla^2\psi\bigr)
$$

The real potential $V$ cancels: $\tfrac{-1}{\mathrm{i}\hbar}V\psi^*\psi + \tfrac{1}{\mathrm{i}\hbar}V\psi^*\psi = 0$.

**Step 2: Identify the divergence.**

The vector identity $\nabla\cdot(f\nabla g) = (\nabla f)\cdot(\nabla g) + f\nabla^2 g$ implies

$$
\psi\nabla^2\psi^* - \psi^*\nabla^2\psi = \nabla\cdot(\psi\nabla\psi^*) - \nabla\cdot(\psi^*\nabla\psi) = \nabla\cdot(\psi\nabla\psi^* - \psi^*\nabla\psi)
$$

Therefore:

$$
\frac{\partial\rho}{\partial t} = -\nabla\cdot\left[\frac{\hbar}{2m\mathrm{i}}(\psi^*\nabla\psi - \psi\nabla\psi^*)\right]
$$

**Step 3: Define the probability current.**

The **probability current density** is

$$
\boldsymbol{j} = \frac{\hbar}{2m\mathrm{i}}\bigl(\psi^*\nabla\psi - \psi\nabla\psi^*\bigr) = \frac{\hbar}{m}\,\mathrm{Im}(\psi^*\nabla\psi)
$$

In the polar form $\psi = R\,\mathrm{e}^{\mathrm{i}S/\hbar}$ (with $R, S$ real), this becomes $\boldsymbol{j} = R^2\,\nabla S/m = \rho\,\boldsymbol{v}$, where $\boldsymbol{v} = \nabla S/m$ is the local flow velocity (classical momentum $p = \nabla S$).

**Step 4: Global conservation.**

The continuity equation $\partial_t\rho + \nabla\cdot\boldsymbol{j} = 0$ integrates to

$$
\frac{\mathrm{d}}{\mathrm{d}t}\int\vert\psi\vert^2\,\mathrm{d}^3x = -\oint_{\text{surface}}\boldsymbol{j}\cdot\mathrm{d}\boldsymbol{A} = 0
$$

(using the divergence theorem and $\psi \to 0$ at infinity). Total probability is conserved for all time.

**Gauge extension.** For a charged particle in a magnetic field, minimal coupling $\hat{\boldsymbol{p}} \to \hat{\boldsymbol{p}} - q\boldsymbol{A}$ modifies the current to the gauge-invariant form:

$$
\boldsymbol{j} = \frac{1}{m}\,\mathrm{Re}\!\left[\psi^*\!\left(\frac{\hbar}{\mathrm{i}}\nabla - q\boldsymbol{A}\right)\!\psi\right] = \frac{\hbar}{m}\,\mathrm{Im}(\psi^*\nabla\psi) - \frac{q}{m}\boldsymbol{A}\vert\psi\vert^2
$$

This equals $\rho\,\boldsymbol{v}_{\rm kin}$ with $m\boldsymbol{v}_{\rm kin} = \boldsymbol{p} - q\boldsymbol{A}$ the kinetic momentum, consistent with the gauge-invariant velocity derived in [4.1.2 Electromagnetic Coupling](../ch4_phase-and-gauge/4-1-2-electromagnetic-coupling).
:::

- Starting from the Schrödinger equation, can we uniquely reconstruct the path integral?
- For time-dependent Hamiltonians $\hat{H}(t)$, how does the path integral formulation change?
- Does the choice of splitting (kinetic vs potential factor ordering) affect the $\delta t \to 0$ limit?

:::

### Summary

- The Schrödinger equation emerges from the path integral via **small-time-step expansion**: each $\delta t$ slice contributes a kinetic Gaussian and a potential phase.
- The three-step derivation shows: (1) Taylor expand the wavefunction, (2) evaluate Gaussian moments, (3) collect $O(\delta t)$ terms.
- The kinetic energy produces the **Laplacian $\frac{\partial^2}{\partial x^2}$**; the potential acts as a **multiplicative factor**.
- The path integral (global, sum over histories) and the Schrödinger equation (local, differential) are **mathematically equivalent**.
- The **propagator** is the Green's function of the Schrödinger equation and solves it with a delta-function source.
