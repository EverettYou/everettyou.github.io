## Lecture Notes

:::{admonition} Definition
:class: important

The **WKB (Wentzel-Kramers-Brillouin) approximation** is a systematic method for solving the Schrödinger equation when the potential varies slowly on the scale of the de Broglie wavelength. The wavefunction's phase and amplitude are treated separately, with phase evolving like the classical action and amplitude diffusing quantum-mechanically. This semi-classical picture is extraordinarily powerful for understanding tunneling, scattering, and bound states when quantum effects are small corrections to classical behavior.
:::

### The WKB Ansatz: From Schrödinger to Hamilton-Jacobi

Assume the wavefunction has the form

$$\psi(x, t) = A(x, t) \, \mathrm{e}^{\mathrm{i}S(x, t)/\hbar}$$


where:
- $S(x, t)$ is the **action function** (related to the classical action)
- $A(x, t)$ is the **amplitude** (slowly-varying)

This is the **WKB ansatz**—a split between rapidly oscillating phase and slowly-varying envelope.

#### Deriving the Equations for $S$ and $A$

:::{admonition} Derivation: WKB Equations from Schrödinger
:class: dropdown note

Substitute into the time-dependent Schrödinger equation:

$$\mathrm{i}\hbar\frac{\partial\psi}{\partial t} = -\frac{\hbar^2}{2m}\frac{\partial^2\psi}{\partial x^2} + V(x)\psi$$


Compute derivatives:

$$\frac{\partial\psi}{\partial t} = \left(\frac{\partial A}{\partial t} + \mathrm{i}\frac{A}{\hbar}\frac{\partial S}{\partial t}\right) \mathrm{e}^{\mathrm{i}S/\hbar}$$


$$\frac{\partial^2\psi}{\partial x^2} = \left(\frac{\partial^2 A}{\partial x^2} + 2\mathrm{i}\frac{1}{\hbar}\frac{\partial A}{\partial x}\frac{\partial S}{\partial x} + \mathrm{i}\frac{A}{\hbar}\frac{\partial^2 S}{\partial x^2} - \frac{A}{\hbar^2}\left(\frac{\partial S}{\partial x}\right)^2\right) \mathrm{e}^{\mathrm{i}S/\hbar}$$


Substituting and canceling $\mathrm{e}^{\mathrm{i}S/\hbar}$, then separating into real and imaginary parts:

**Imaginary part** (leading order, dominates for large $S/\hbar$):

$$\frac{\partial S}{\partial t} + \frac{1}{2m}\left(\frac{\partial S}{\partial x}\right)^2 + V(x) = 0$$


This is the **Hamilton-Jacobi equation**, the classical equation for the action!

**Real part** (higher order, corrects the amplitude):

$$\frac{\partial A}{\partial t} + \frac{1}{m}\frac{\partial A}{\partial x}\frac{\partial S}{\partial x} + \frac{A}{2m}\frac{\partial^2 S}{\partial x^2} = \frac{\hbar^2}{2im}\frac{\partial^2 A}{\partial x^2}$$


In the limit $\hbar \to 0$, the right side vanishes and we get a **continuity equation** for the amplitude.

:::

### Time-Independent Schrödinger Equation in WKB

For a **stationary state** with energy $E$, write $\psi(x, t) = \phi(x) \mathrm{e}^{-\mathrm{i}Et/\hbar}$ and $S(x, t) = S_0(x) - Et$. The Hamilton-Jacobi equation becomes:

$$E = \frac{1}{2m}\left(\frac{\mathrm{d}S_0}{dx}\right)^2 + V(x)$$


Rearranging:

$$\frac{\mathrm{d}S_0}{dx} = \pm \sqrt{2m(E - V(x))} = \pm p(x)$$


where $p(x) = \sqrt{2m(E - V(x))}$ is the **classical momentum**.

#### Solution in Classically Allowed Region

In a region where $E > V(x)$ (classically allowed), the momentum is real:

$$\phi(x) \propto \frac{1}{\sqrt{p(x)}} \exp\left(\pm\frac{\mathrm{i}}{\hbar}\int_0^x p(x') dx'\right)$$


The $\pm$ corresponds to waves traveling right ($+$) or left ($-$). The amplitude $1/\sqrt{p}$ reflects **probability conservation**: where the momentum is small (slow particle), the wavefunction is large (must stay in the region longer).

#### Solution in Classically Forbidden Region

In a region where $E < V(x)$ (classically forbidden), the momentum is imaginary: $p \to \mathrm{i}\kappa$ where $\kappa(x) = \sqrt{2m(V(x) - E)}$. Then:

$$\phi(x) \propto \frac{1}{\sqrt{\kappa(x)}} \exp\left(\mp\frac{1}{\hbar}\int_0^x \kappa(x') dx'\right)$$


The wavefunction **decays exponentially** into the classically forbidden region. The characteristic penetration depth is

$$\lambda_\text{decay} = \frac{\hbar}{\kappa} = \frac{\hbar}{\sqrt{2m(V - E)}}$$


### Connection Formulas: Matching at Turning Points

A **classical turning point** occurs where $E = V(x)$, so $p = 0$. At such points, the WKB solutions diverge ($\sim 1/\sqrt{p}$) and must be matched carefully.

:::{admonition} Discussion: WKB Exactness and Semi-Classical Limits
:class: dropdown warning

The WKB approximation treats $\hbar$ as small and expands the wavefunction in powers of $\hbar$. Yet in many applications, WKB predictions match experiments well even when $\hbar$ is not truly small. This prompts important questions:

- For any *specific* potential, are there energy ranges where WKB becomes exact (zero error)? For example, is WKB exact for free particles, or only for certain potentials?
- WKB fails near turning points where $p(x) = \sqrt{2m(E-V)} = 0$. Connection formulas repair this failure. But do connection formulas assume anything about the shape of the potential near the turning point?
- For barrier tunneling, WKB predicts the transmission coefficient via $T \propto \mathrm{e}^{-2\int p \, \mathrm{d}x/\hbar}$. How large can the error be? Can you design a potential where WKB is very wrong?
- Why does WKB work so well for atomic structure (predicting energy levels) when $\hbar$ is definitely not small? Is there a hidden small parameter in atoms?
:::
#### Simple Turning Point

Near a simple turning point $x_0$ where $V(x) \approx V(x_0) + V'(x_0)(x - x_0)$:

$$p(x) \approx \sqrt{2mV'(x_0)} \sqrt{x - x_0}$$


The WKB solution oscillates in the allowed region ($x < x_0$) and decays in the forbidden region ($x > x_0$). The **connection formulas** relate the oscillating solution on one side to the decaying solution on the other:

**Left of turning point** ($x < x_0$, allowed region):

$$\psi(x) \approx \frac{C}{\sqrt{p(x)}} \sin\left(\frac{1}{\hbar}\int_x^{x_0} p(x') dx' + \frac{\pi}{4}\right)$$


**Right of turning point** ($x > x_0$, forbidden region):

$$\psi(x) \approx \frac{C}{\sqrt{\kappa(x)}} \exp\left(-\frac{1}{\hbar}\int_{x_0}^x \kappa(x') dx'\right)$$


The phase $\pi/4$ is crucial—it accounts for the Airy function behavior near the turning point.

### Tunneling Through a Barrier

Consider a particle encountering a potential barrier of width $w$ and height $V_0 > E$. The WKB approximation gives the **tunneling probability**:

$$T \approx \mathrm{e}^{-2\gamma}, \quad \gamma = \frac{1}{\hbar}\int_a^b \kappa(x) dx$$


where the integral is from the left turning point $a$ to the right turning point $b$.

**For a square barrier** ($V(x) = V_0$ for $0 < x < w$, $V = 0$ elsewhere):

$$\gamma = \frac{w}{\hbar}\sqrt{2m(V_0 - E)}$$


$$T \approx \exp\left(-\frac{2w}{\hbar}\sqrt{2m(V_0 - E)}\right)$$


:::{admonition} Example: Tunneling Rate
:class: example

For an electron tunneling through a 1 nm barrier with height 1 eV above the ground state energy:

$$\gamma \approx (10^{-9} \text{ m})\sqrt{2(10^{-30} \text{ kg})(1.6 \times 10^{-19} \text{ J})} / (10^{-34} \text{ J·s}) \sim 5$$


So $T \sim \mathrm{e}^{-10} \sim 10^{-5}$, meaning about 1 in $10^5$ attempts tunnel through. This is why quantum tunneling is rare but non-zero for macroscopic barriers, yet essential for atomic and nuclear processes.
:::

### Validity of WKB

The WKB approximation is valid when the potential varies slowly on the scale of the de Broglie wavelength:

$$\left|\frac{\mathrm{d}V}{dx}\right| \ll \frac{p^2}{m} = \frac{2m(E - V)}{m}$$


Or equivalently:

$$\left|\frac{\mathrm{d}\ln p}{dx}\right| = \left|\frac{1}{p}\frac{\mathrm{d}V}{dx}\right| \ll 1$$


This breaks down near:
- **Turning points** where $p \to 0$ (addressed by connection formulas)
- **Potential discontinuities** (sharp steps)
- **Rapidly oscillating potentials** (wavelength comparable to particle wavelength)

For most smooth potentials, WKB accuracy is a few percent or better.

:::{admonition} Bohr-Sommerfeld Quantization
:class: important

For a particle bound in a potential well with classical turning points at $x = a$ and $x = b$ (where $V(a) = V(b) = E$), the connection formulas at both turning points yield the **Bohr-Sommerfeld quantization condition**:

$$\int_a^b p(x) dx = \pi\hbar\left(n + \frac{1}{2}\right), \quad n = 0, 1, 2, \ldots$$


The factor $1/2$ comes from the phase shifts at each turning point (each contributes $\pi/4$ at entry and exit).

With $p(x) = \sqrt{2m(E - V(x))}$, this determines the **allowed energy levels** $E_n$ for a bound state.
:::

:::{admonition} Derivation: Bohr-Sommerfeld from Connection Formulas
:class: dropdown note

For a **closed orbit** (particle bound between $a$ and $b$), we must apply connection formulas at both turning points and require that the wavefunction remains single-valued.

**In the allowed region** ($a < x < b$), the WKB wavefunction oscillates:

$$\phi(x) = \frac{C}{\sqrt{p(x)}} \sin\left(\frac{1}{\hbar}\int_a^x p(x') dx' + \frac{\pi}{4}\right)$$


For a **closed orbit**, the phase accumulated going from $a$ to $b$ and back to $a$ must be a multiple of $2\pi$:

$$2 \int_a^b p(x) dx = 2\pi\hbar\left(n + \frac{1}{2}\right)$$


where $n = 0, 1, 2, \ldots$ is the quantum number. Simplifying:

$$\int_a^b p(x) dx = \pi\hbar\left(n + \frac{1}{2}\right)$$


This condition determines the allowed energy levels $E_n$.

**Example: Harmonic Oscillator**

For $V(x) = \frac{m\omega^2}{2}x^2$, the integral yields

$$E_n = \hbar\omega\left(n + \frac{1}{2}\right)$$


which is **exactly the quantum result**—WKB gives the exact answer for the harmonic oscillator despite being an approximation in general.

:::

:::{admonition} Note: Classical-Quantum Correspondence
:class: note

The transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the **correspondence principle**.

In quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::

### Summary

- **WKB ansatz**: $\psi = A \mathrm{e}^{\mathrm{i}S/\hbar}$ separates phase from amplitude
- **Phase equation**: $S$ obeys the Hamilton-Jacobi equation—classical mechanics emerges
- **Momentum and wavelength**: $p = \partial S/\partial x$; WKB wavelength is local ($\lambda = 2\pi\hbar/p(x)$)
- **Allowed region**: Wavefunction oscillates as $1/\sqrt{p} \sin(\int p dx/\hbar + \text{phase})$
- **Forbidden region**: Wavefunction decays exponentially as $1/\sqrt{\kappa} \exp(-\int \kappa dx/\hbar)$
- **Connection formulas**: Match oscillating and decaying solutions across turning points with proper phase shifts
- **Tunneling**: Probability $T \approx \exp(-2\int\kappa dx/\hbar)$ is exponentially suppressed by barrier width and height
- **Bohr-Sommerfeld quantization**: Determines bound-state energy levels through $\int_a^b p(x) dx = \pi\hbar(n + 1/2)$
- **Validity**: Excellent for slowly-varying potentials; breaks down near turning points (remedied by connection formulas) and sharp features
