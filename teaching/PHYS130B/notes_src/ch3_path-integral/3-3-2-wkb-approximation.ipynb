## Lecture Notes

The WKB approximation treats the wavefunction's phase and amplitude separately, with phase evolving like the classical action and amplitude diffusing quantum-mechanically. This semi-classical picture is extraordinarily powerful for understanding tunneling, scattering, and bound states when quantum effects are small corrections to classical behavior.

### The WKB Ansatz: From Schrödinger to Hamilton-JacobiThe **WKB (Wentzel-Kramers-Brillouin) approximation** is a systematic method for solving the Schrödinger equation when the potential varies slowly on the scale of the de Broglie wavelength.

#### WKB AnsatzAssume the wavefunction has the form

$$\psi(x, t) = A(x, t) \, e^{iS(x, t)/\hbar}

$$ (eq-wkb-ansatz)

where:- $S(x, t)$ is the **action function** (related to the classical action).- $A(x, t)$ is the **amplitude** (slowly-varying).This is the **WKB ansatz**—a split between rapidly oscillating phase and slowly-varying envelope.

#### Deriving the Equations for $S$ and $A$:::{admonition} Derivation: WKB Equations from Schrödinger
:class: dropdown note

Substitute into the time-dependent Schrödinger equation:

$$i\hbar\frac{\partial\psi}{\partial t} = -\frac{\hbar^2}{2m}\frac{\partial^2\psi}{\partial x^2} + V(x)\psi

$$Compute derivatives:

$$\frac{\partial\psi}{\partial t} = \left(\frac{\partial A}{\partial t} + i\frac{A}{\hbar}\frac{\partial S}{\partial t}\right) e^{iS/\hbar}

$$$$\frac{\partial^2\psi}{\partial x^2} = \left(\frac{\partial^2 A}{\partial x^2} + 2i\frac{1}{\hbar}\frac{\partial A}{\partial x}\frac{\partial S}{\partial x} + i\frac{A}{\hbar}\frac{\partial^2 S}{\partial x^2} - \frac{A}{\hbar^2}\left(\frac{\partial S}{\partial x}\right)^2\right) e^{iS/\hbar}

$$Substituting and canceling $e^{iS/\hbar}$:

$$i\hbar\frac{\partial A}{\partial t} - A\frac{\partial S}{\partial t} = -\frac{\hbar^2}{2m}\frac{\partial^2 A}{\partial x^2} - i\frac{\hbar}{m}\frac{\partial A}{\partial x}\frac{\partial S}{\partial x} - i\frac{A}{2m}\frac{\partial^2 S}{\partial x^2} + \frac{A}{2m}\left(\frac{\partial S}{\partial x}\right)^2 + VA

$$Separate into **imaginary** and **real** parts:**Imaginary part** (leading order, dominates for large $S/\hbar$):

$$\frac{\partial S}{\partial t} + \frac{1}{2m}\left(\frac{\partial S}{\partial x}\right)^2 + V(x) = 0

$$ (eq-hamilton-jacobi-wkb)

This is the **Hamilton-Jacobi equation**, the classical equation for the action!**Real part** (higher order, corrects the amplitude):

$$\frac{\partial A}{\partial t} + \frac{1}{m}\frac{\partial A}{\partial x}\frac{\partial S}{\partial x} + \frac{A}{2m}\frac{\partial^2 S}{\partial x^2} = \frac{\hbar^2}{2im}\frac{\partial^2 A}{\partial x^2}

$$ (eq-amplitude-equation)

In the limit $\hbar \to 0$, the right side vanishes and we get a **continuity equation** for the amplitude.

:::



### Time-Independent Schrödinger Equation in WKBFor a **stationary state** with energy $E$, write $\psi(x, t) = \phi(x) e^{-iEt/\hbar}$ and $S(x, t) = S_0(x) - Et$. The Hamilton-Jacobi equation becomes:

$$E = \frac{1}{2m}\left(\frac{dS_0}{dx}\right)^2 + V(x)

$$ (eq-wkb-spatial)

Rearranging:

$$\frac{dS_0}{dx} = \pm \sqrt{2m(E - V(x))} = \pm p(x)

$$ (eq-momentum-def)

where $p(x) = \sqrt{2m(E - V(x))}$ is the **classical momentum**.

#### Solution in Classically Allowed RegionIn a region where $E > V(x)$ (classically allowed), the momentum is real:

$$\phi(x) \propto \frac{1}{\sqrt{p(x)}} \exp\left(\pm\frac{i}{\hbar}\int_0^x p(x') dx'\right)

$$ (eq-wkb-allowed)

The $\pm$ corresponds to waves traveling right ($+$) or left ($-$). The amplitude $1/\sqrt{p}$ reflects **probability conservation**: where the momentum is small (slow particle), the wavefunction is large (must stay in the region longer).

#### Solution in Classically Forbidden RegionIn a region where $E < V(x)$ (classically forbidden), the momentum is imaginary: $p \to i\kappa$ where $\kappa(x) = \sqrt{2m(V(x) - E)}$. Then:

$$\phi(x) \propto \frac{1}{\sqrt{\kappa(x)}} \exp\left(\mp\frac{1}{\hbar}\int_0^x \kappa(x') dx'\right)

$$ (eq-wkb-forbidden)

The wavefunction **decays exponentially** into the classically forbidden region. The characteristic penetration depth is

$$\lambda_\text{decay} = \frac{\hbar}{\kappa} = \frac{\hbar}{\sqrt{2m(V - E)}}

$$ (eq-decay-length)

### Connection Formulas: Matching at Turning PointsA **classical turning point** occurs where $E = V(x)$, so $p = 0$. At such points, the WKB solutions diverge ($\sim 1/\sqrt{p}$) and must be matched carefully.

#### Simple Turning PointNear a simple turning point $x_0$ where $V(x) \approx V(x_0) + V'(x_0)(x - x_0)$:

$$p(x) \approx \sqrt{2mV'(x_0)} \sqrt{x - x_0}

$$The WKB solution oscillates in the allowed region ($x < x_0$) and decays in the forbidden region ($x > x_0$). The **connection formulas** relate the oscillating solution on one side to the decaying solution on the other:**Left of turning point** ($x < x_0$, allowed region):

$$\psi(x) \approx \frac{C}{\sqrt{p(x)}} \sin\left(\frac{1}{\hbar}\int_x^{x_0} p(x') dx' + \frac{\pi}{4}\right)

$$ (eq-left-turning)

**Right of turning point** ($x > x_0$, forbidden region):

$$\psi(x) \approx \frac{C}{\sqrt{\kappa(x)}} \exp\left(-\frac{1}{\hbar}\int_{x_0}^x \kappa(x') dx'\right)

$$ (eq-right-turning)

The phase $\pi/4$ is crucial—it accounts for the Airy function behavior near the turning point.

### Tunneling Through a BarrierConsider a particle encountering a potential barrier of width $w$ and height $V_0 > E$. The WKB approximation gives the **tunneling probability**:

$$T \approx e^{-2\gamma}, \quad \gamma = \frac{1}{\hbar}\int_a^b \kappa(x) dx

$$ (eq-tunneling-probability)

where the integral is from the left turning point $a$ to the right turning point $b$.**For a square barrier** ($V(x) = V_0$ for $0 < x < w$, $V = 0$ elsewhere):

$$\gamma = \frac{w}{\hbar}\sqrt{2m(V_0 - E)}

$$$$T \approx \exp\left(-\frac{2w}{\hbar}\sqrt{2m(V_0 - E)}\right)

$$ (eq-square-barrier)

:::{admonition} Tunneling rate:class: exampleFor an electron tunneling through a 1 nm barrier with height 1 eV above the ground state energy:

$$\gamma \approx (10^{-9} \text{ m})\sqrt{2(10^{-30} \text{ kg})(1.6 \times 10^{-19} \text{ J})} / (10^{-34} \text{ J·s}) \sim 5

$$So $T \sim e^{-10} \sim 10^{-5}$, meaning about 1 in $10^5$ attempts tunnel through. This is why quantum tunneling is rare but non-zero for macroscopic barriers, yet essential for atomic and nuclear processes.
:::
### Validity of WKBThe WKB approximation is valid when the potential varies slowly on the scale of the de Broglie wavelength:

$$\left|\frac{dV}{dx}\right| \ll \frac{p^2}{m} = \frac{2m(E - V)}{m}

$$ (eq-wkb-validity)

Or equivalently:

$$\left|\frac{d\ln p}{dx}\right| = \left|\frac{1}{p}\frac{dV}{dx}\right| \ll 1

$$ (eq-wkb-validity-alt)

This breaks down near:
- **Turning points** where $p \to 0$ (addressed by connection formulas).
- **Potential discontinuities** (sharp steps).
- **Rapidly oscillating potentials** (wavelength comparable to particle wavelength).For most smooth potentials, WKB accuracy is a few percent or better.

:::{admonition} Classical-quantum correspondence:class: seealsoThe transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the correspondence principle.
:::
:::{admonition} Action and phase:class: noteIn quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.


:::{admonition} Derivation: WKB Quantization Condition from Connection Formulas
:class: dropdown note

For a particle bound in a potential well with classical turning points at $x = a$ and $x = b$ (where $E = V(a) = V(b)$), we must apply connection formulas at both turning points and require that the wavefunction remains single-valued (periodic) around a closed loop.

**In the allowed region** ($a < x < b$), the WKB wavefunction oscillates:

$$\phi(x) = \frac{C}{\sqrt{p(x)}} \sin\left(\frac{1}{\hbar}\int_a^x p(x') dx' + \frac{\pi}{4}\right)

$$

**At the right turning point** ($x = b$), the connection formula relates this to the decaying solution in the forbidden region ($x > b$).

**At the left turning point** ($x = a$), the connection formula relates the oscillating solution to a decaying solution in the forbidden region ($x < a$).

For a **closed orbit** (particle bound between $a$ and $b$), the phase accumulated going from $a$ to $b$ and back to $a$ must be a multiple of $2\pi$:

$$2 \int_a^b p(x) dx = 2\pi\hbar\left(n + \frac{1}{2}\right)

$$where $n = 0, 1, 2, \ldots$ is the quantum number. The factor of $1/2$ comes from the phase shifts at each turning point (each contributes $\pi/4$ at entry and exit).

Simplifying:

$$\int_a^b p(x) dx = \pi\hbar\left(n + \frac{1}{2}\right)

$$

With $p(x) = \sqrt{2m(E - V(x))}$, this becomes the **Bohr-Sommerfeld quantization condition**:

$$\int_a^b \sqrt{2m(E_n - V(x))} \, dx = \pi\hbar\left(n + \frac{1}{2}\right)

$$

This condition determines the **allowed energy levels** $E_n$ for a bound state. For large $n$ (highly excited states), $E_n \gg V_\text{min}$, and the formula reduces to the classical action integral.

**Example: Harmonic Oscillator**

For $V(x) = \frac{m\omega^2}{2}x^2$, the integral yields

$$E_n = \hbar\omega\left(n + \frac{1}{2}\right)

$$which is **exactly the quantum result**—WKB gives the exact answer for the harmonic oscillator despite being an approximation in general.

### Summary
- **WKB ansatz**: $\psi = A e^{iS/\hbar}$ separates phase from amplitude.
- **Phase equation**: $S$ obeys the Hamilton-Jacobi equation—classical mechanics emerges.
- **Momentum and wavelength**: $p = \partial S/\partial x$; WKB wavelength is local ($\lambda = 2\pi\hbar/p(x)$).
- **Allowed region**: Wavefunction oscillates as $1/\sqrt{p} \sin(\int p dx/\hbar + \text{phase})$.
- **Forbidden region**: Wavefunction decays exponentially as $1/\sqrt{\kappa} \exp(-\int \kappa dx/\hbar)$.
- **Connection formulas**: Match oscillating and decaying solutions across turning points with proper phase shifts.
- **Tunneling**: Probability $T \approx \exp(-2\int\kappa dx/\hbar)$ is exponentially suppressed by barrier width and height.
- **Validity**: Excellent for slowly-varying potentials; breaks down near turning points (remedied by connection formulas) and sharp features.
