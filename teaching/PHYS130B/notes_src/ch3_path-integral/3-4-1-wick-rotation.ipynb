## Lecture Notes

The Wick rotation from real time $t$ to imaginary time $\tau = it/\hbar$ transforms the oscillatory quantum path integral into an exponentially damped sum. This analytic continuation reveals that quantum mechanics and thermal statistical mechanics are intimately related—they are two faces of the same mathematical structure.

### The Wick RotationIn quantum field theory and statistical mechanics, analytic continuation from real time $t$ to **imaginary time** $\tau = -it/\hbar$ (or $t \to it$ in other conventions) often simplifies calculations.

#### DefinitionThe **Wick rotation** is the transformation

$$t \to -i\tau, \quad \text{where} \quad \tau \in \mathbb{R}

$$ (eq-wick-rotation)

Equivalently, $\tau = it/\hbar$ is the **imaginary time**, with dimension of time. The range $t \in [0, T]$ (real time) maps to $\tau \in [0, -iT]$, but it's more convenient to write $\tau \in [0, \beta\hbar]$ where $\beta = 1/(k_B T)$ is the **inverse temperature** (if thermal context applies).

#### Effect on the Action:::{admonition} Derivation: Wick Rotation Transform
:class: dropdown note

The **real-time action** is

$$S[\boldsymbol{x}] = \int_0^T \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})\right) dt

$$ (eq-real-action)

Under $t \to -i\tau$, we have $dt \to -i d\tau$ and $\dot{\boldsymbol{x}} = d\boldsymbol{x}/dt \to d\boldsymbol{x}/(-i d\tau) = i d\boldsymbol{x}/d\tau$. Thus:

$$S[\boldsymbol{x}] \to \int_0^{\beta\hbar} \left(\frac{m}{2}(i\dot{\boldsymbol{x}})^2 - V(\boldsymbol{x})\right) (-i) d\tau = -i \int_0^{\beta\hbar} \left(-\frac{m}{2}\dot{\boldsymbol{x}}^2 - V(\boldsymbol{x})\right) d\tau

$$The minus signs combine to give:

$$S[\boldsymbol{x}] \to -i(-i) \int_0^{\beta\hbar} \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 + V(\boldsymbol{x})\right) d\tau = -\int_0^{\beta\hbar} \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 + V(\boldsymbol{x})\right) d\tau

$$Defining the **Euclidean action** as

$$S_E[\boldsymbol{x}] = \int_0^{\beta\hbar} \left(\frac{m}{2}\dot{\boldsymbol{x}}^2 + V(\boldsymbol{x})\right) d\tau

$$ (eq-euclidean-action)we have $S \to -iS_E$, so

:::



$$e^{iS/\hbar} \to e^{-S_E/\hbar}

$$ (eq-phase-to-exponential)

The **rapidly oscillating phase** $e^{iS/\hbar}$ becomes a **rapidly decaying exponential** $e^{-S_E/\hbar}$. This convergence is crucial for numerical simulations and analytical approximations.

### Euclidean Path IntegralThe **Euclidean path integral** is

$$Z(\beta) = \int_{\text{periodic}} e^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$ (eq-euclidean-path-integral)

where the integral is over paths that are **periodic in imaginary time**: $\boldsymbol{x}(\tau + \beta\hbar) = \boldsymbol{x}(\tau)$.**Key properties**:1. **Convergence**: The exponential $e^{-S_E/\hbar}$ decays exponentially for large actions. Paths with large Euclidean action are suppressed. This makes numerical simulation feasible.2. **Positive measure**: Unlike the oscillating real-time path integral, the Euclidean integral has a well-defined probability measure (positive definite). This is essential for Monte Carlo simulations.3. **Periodicity**: The periodicity in imaginary time $\tau$ reflects the **cyclic nature** of the thermal density matrix in quantum statistical mechanics.

### Connection to the Partition FunctionIn quantum statistical mechanics, the **partition function** at temperature $T$ is

$$Z(T) = \text{Tr}(e^{-\beta H}), \quad \beta = \frac{1}{k_B T}

$$ (eq-partition-trace)

where $H$ is the Hamiltonian and the trace sums over all quantum states.**Path integral representation**: The partition function can be written as

$$Z(\beta) = \int_{\text{periodic}} e^{-S_E[\boldsymbol{x}]/\hbar} \, \mathcal{D}[\boldsymbol{x}]

$$ (eq-partition-path-integral)

where the sum is over all paths periodic in imaginary time with period $\beta\hbar = 1/(k_B T)$.

:::{admonition} Path integral = thermal trace:class: importantThe partition function $Z = \text{Tr}(e^{-\beta H})$ equals the functional integral over periodic paths in Euclidean time. This is a profound connection:
- **Trace**: sums over all eigenstates.
- **Path integral**: sums over all closed paths in imaginary time.They give the same answer, which is a testament to the power and universality of the path integral formulation.
:::
### Analytic ContinuationResults computed in imaginary time (Euclidean path integral) can be analytically continued back to real time through various methods:

#### Direct ContinuationIf a result is known as a function $F(\tau)$ on the imaginary-time axis, it can often be extended to real time $t$ by replacing $\tau \to it$:

$$F(t) = F(\tau)|_{\tau \to it}

$$ (eq-analytic-continuation)

This works if $F$ is analytic in the complex-$\tau$ plane.

#### Fourier Transform ContinuationFor correlators or Green's functions, expand in **Matsubara frequencies**:

$$G(\tau) = \sum_{n=-\infty}^{\infty} \frac{1}{\beta\hbar} \tilde{G}(i\omega_n) e^{-i\omega_n \tau}

$$where $\omega_n = 2\pi n/(\beta\hbar)$ are the **Matsubara frequencies** (quantized imaginary frequencies). Continuing to real frequency $\omega$ gives the real-time result:

$$G(t) \leftrightarrow \tilde{G}(\omega)

$$ (eq-matsubara)

### Physical Interpretation

#### Low Temperature LimitAs $T \to 0$, we have $\beta \to \infty$. The period of the imaginary-time path becomes infinite. Only the **ground state** (lowest energy) contributes significantly to the path integral:

$$Z(T \to 0) \approx e^{-\beta E_0}

$$where $E_0$ is the ground-state energy. Thus, **zero-temperature path integrals govern ground-state properties**.

#### High Temperature LimitAs $T \to \infty$, $\beta \to 0$. Many paths contribute, and the quantum kinetic energy term $\propto \dot{\boldsymbol{x}}^2$ becomes negligible. The partition function approaches the **classical statistical mechanics result**:

$$Z(T \to \infty) \approx \int e^{-V(\boldsymbol{x})/(k_B T)} d^3\boldsymbol{x}

$$ (eq-classical-limit)

This is the **classical partition function**, weighted only by Boltzmann factors for potential energy. Quantum effects vanish at high temperature.

:::{admonition} Classical-quantum correspondence:class: seealsoThe transition from quantum to classical mechanics is smooth. When the action $S$ is much larger than $\hbar$, the phase varies rapidly and only the stationary-action path survives quantum interference. This is the essence of the correspondence principle.
:::
:::{admonition} Action and phase:class: noteIn quantum mechanics, the action $S$ plays the role that phase $\phi = S/\hbar$ plays in waves. The Planck constant $\hbar$ sets the scale: when $S \sim \hbar$, quantum effects are significant; when $S \gg \hbar$, classical behavior dominates.
:::
### Summary
- **Wick rotation**: $t \to -i\tau$ transforms real-time oscillating path integral to imaginary-time converging one.
- **Euclidean action**: $S_E = \int (\frac{m}{2}\dot{\boldsymbol{x}}^2 + V) d\tau$; has no kinetic term with opposite sign.
- **Phase to exponential**: $e^{iS/\hbar} \to e^{-S_E/\hbar}$ makes path integral numerically tractable.
- **Partition function**: Euclidean path integral over periodic paths equals $Z = \text{Tr}(e^{-\beta H})$.
- **Periodicity**: Imaginary-time period $\beta\hbar = 1/(k_B T)$ encodes temperature.
- **Analytic continuation**: Results from imaginary time can be continued to real time via analytic functions or Matsubara frequencies.
- **Temperature limits**: $T \to 0$ selects ground state; $T \to \infty$ recovers classical mechanics.
