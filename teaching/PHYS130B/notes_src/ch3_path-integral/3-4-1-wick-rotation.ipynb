## Lecture Notes

### Overview

The path integral oscillates as $\mathrm{e}^{\mathrm{i}S/\hbar}$, making it difficult to evaluate directly. The **Wick rotation** $t \to -\mathrm{i}\tau$ replaces oscillating phases with exponentially decaying weights $\mathrm{e}^{-S_E/\hbar}$, where $S_E$ is the **Euclidean action**. This transform reveals a deep connection: the imaginary-time propagator for interval $\beta\hbar$ is precisely the thermal density matrix $\mathrm{e}^{-\beta \hat{H}}$, linking quantum dynamics to statistical mechanics.

:::{admonition} Wick Rotation
:class: important

The **Wick rotation** from real time $t$ to imaginary time $\tau = it/\hbar$ transforms the oscillatory quantum path integral into an exponentially damped sum. This analytic continuation reveals that quantum mechanics and thermal statistical mechanics are intimately related—they are two faces of the same mathematical structure.
:::

### The Wick Rotation

In quantum field theory and statistical mechanics, analytic continuation from real time $t$ to **imaginary time** $\tau = -it/\hbar$ simplifies calculations.

#### Definition

The **Wick rotation** is the transformation

$$t \to -\mathrm{i}\tau, \quad \text{where} \quad \tau \in \mathbb{R}$$


Equivalently, $\tau = it/\hbar$ is the **imaginary time**, with dimension of time. If thermal context applies, write $\tau \in [0, \beta\hbar]$ where $\beta = 1/(k_B T)$ is the **inverse temperature**.

:::{admonition} Derivation: Wick Rotation Transform
:class: dropdown information

The **real-time action** is

$$S[x] = \int_0^T \left(\frac{m}{2}\dot{x}^2 - V(x)\right) dt$$


Under $t \to -\mathrm{i}\tau$, we have $dt \to -i d\tau$ and $\dot{x} = dx/dt \to i dx/d\tau$. Thus:

$$S[x] \to \int_0^{\beta\hbar} \left(\frac{m}{2}(i\dot{x})^2 - V(x)\right) (-i) d\tau = -i \int_0^{\beta\hbar} \left(-\frac{m}{2}\dot{x}^2 - V(x)\right) d\tau$$


The minus signs combine to give:

$$S[x] \to -\int_0^{\beta\hbar} \left(\frac{m}{2}\dot{x}^2 + V(x)\right) d\tau$$


Defining the **Euclidean action** as

$$S_E[x] = \int_0^{\beta\hbar} \left(\frac{m}{2}\dot{x}^2 + V(x)\right) d\tau$$


we have $S \to -iS_E$, so

$$\mathrm{e}^{\mathrm{i}S/\hbar} \to \mathrm{e}^{-S_E/\hbar}$$


:::

The **rapidly oscillating phase** $\mathrm{e}^{\mathrm{i}S/\hbar}$ becomes a **rapidly decaying exponential** $\mathrm{e}^{-S_E/\hbar}$. This convergence is crucial for numerical simulations and analytical approximations.

:::{admonition} Derivation: Euclidean Equations of Motion
:class: dropdown information

The Euclidean action is $S_E[x] = \int_0^{\beta\hbar}\left(\frac{m}{2}\dot{x}^2 + V(x)\right)\mathrm{d}\tau$, where $\dot{x} = \mathrm{d}x/\mathrm{d}\tau$. The classical Euclidean path extremizes $S_E$.

**Euler-Lagrange equation.** Varying $S_E$ with respect to $x(\tau)$:

$$
\frac{\delta S_E}{\delta x} = 0 \implies m\ddot{x} = +\frac{\mathrm{d}V}{\mathrm{d}x}
$$

Compare with the real-time equation $m\ddot{x} = -\mathrm{d}V/\mathrm{d}x$ (Newton's second law). The Euclidean equation of motion describes a particle moving in the **inverted potential** $-V(x)$.

**Physical interpretation.** Under $t \to -\mathrm{i}\tau$, the kinetic energy is unchanged but the force reverses sign. A particle that oscillates in a potential well in real time instead rolls over the inverted barrier in Euclidean time. This is the mechanism behind tunneling: the Euclidean trajectory that crosses a barrier corresponds to a classically forbidden real-time process.

**Energy conservation.** The Euclidean "energy" (conserved along Euclidean trajectories) is

$$
\mathcal{E}_E = \frac{m}{2}\dot{x}^2 - V(x)
$$

Note the minus sign before $V$ — this is the Hamiltonian for the inverted potential. For a tunneling solution starting and ending at rest ($\dot{x} = 0$) at the same potential height, $\mathcal{E}_E = -V(x_0)$ where $x_0$ is the starting point.
:::

:::{admonition} Why Wick Rotation Matters
:class: attention

Real-time path integrals with $\mathrm{e}^{\mathrm{i}S/\hbar}$ are **highly oscillatory**—contributions from different paths interfere constructively and destructively, making direct integration intractable. 

After Wick rotation, the Euclidean path integral with $\mathrm{e}^{-S_E/\hbar}$ is **exponentially damped**—contributions from paths with large action are exponentially suppressed, enabling:
- Numerical Monte Carlo simulations
- Perturbative expansions
- Semi-classical approximations
:::

### Euclidean Path Integral

The **Euclidean path integral** is

$$Z(\beta) = \int_{\text{periodic}} \mathrm{e}^{-S_E[x]/\hbar} \, \mathcal{D}[x]$$


where the integral is over paths that are **periodic in imaginary time**: $x(\tau + \beta\hbar) = x(\tau)$.

**Key properties**:

1. **Convergence**: The exponential $\mathrm{e}^{-S_E/\hbar}$ decays exponentially for large actions. Paths with large Euclidean action are suppressed. This makes numerical simulation feasible.

2. **Positive measure**: Unlike the oscillating real-time path integral, the Euclidean integral has a well-defined probability measure (positive definite). This is essential for Monte Carlo simulations.

3. **Periodicity**: The periodicity in imaginary time $\tau$ reflects the **cyclic nature** of the thermal density matrix in quantum statistical mechanics.

:::{admonition} Derivation: Free Particle Euclidean Propagator
:class: dropdown information

For a free particle ($V = 0$), the Euclidean propagator $K_E(x', \tau; x, 0) = \langle x' | \mathrm{e}^{-\hat{H}\tau/\hbar} | x \rangle$ can be computed exactly.

**Step 1: Classical Euclidean action.**
The Euclidean equation of motion $m\ddot{x} = 0$ gives $x_{\mathrm{cl}}(\tau') = x + (x' - x)\tau'/\tau$. The classical Euclidean action is:

$$
S_E^{\mathrm{cl}} = \int_0^{\tau} \frac{m}{2}\left(\frac{x'-x}{\tau}\right)^2 \mathrm{d}\tau' = \frac{m(x'-x)^2}{2\tau}
$$

**Step 2: Fluctuation integral.**
Writing $x(\tau') = x_{\mathrm{cl}}(\tau') + y(\tau')$ with $y(0) = y(\tau) = 0$, the action separates: $S_E = S_E^{\mathrm{cl}} + S_E^{\mathrm{fluct}}$. The fluctuation integral is a Gaussian over $y(\tau')$ and gives a $\tau$-dependent prefactor.

**Step 3: Result.**
Combining (or equivalently, inserting a complete set of momentum eigenstates $\langle x' | \mathrm{e}^{-\hat{p}^2\tau/(2m\hbar)} | x \rangle$):

$$
K_E(x', \tau; x, 0) = \sqrt{\frac{m}{2\pi\hbar\tau}} \exp\left(-\frac{m(x'-x)^2}{2\hbar\tau}\right)
$$

**Comparison with real time.** The real-time propagator $K(x',t;x,0) = \sqrt{\frac{m}{2\pi\mathrm{i}\hbar t}}\exp\left(\frac{\mathrm{i}m(x'-x)^2}{2\hbar t}\right)$ is obtained by the replacement $\tau \to \mathrm{i}t$, confirming the analytic continuation. The Euclidean propagator is a Gaussian that decays with distance — a **heat kernel** — while the real-time propagator oscillates.
:::

### Connection to the Partition Function

In quantum statistical mechanics, the **partition function** at temperature $T$ is

$$Z(T) = \text{Tr}(\mathrm{e}^{-\beta \hat{H}}), \quad \beta = \frac{1}{k_B T}$$


where $\hat{H}$ is the Hamiltonian and the trace sums over all quantum states.

:::{admonition} Path Integral Representation of Partition Function
:class: important

$$Z(\beta) = \int_{\text{periodic}} \mathrm{e}^{-S_E[x]/\hbar} \, \mathcal{D}[x]$$


where the sum is over all paths periodic in imaginary time with period $\beta\hbar = 1/(k_B T)$.

The partition function $Z = \text{Tr}(\mathrm{e}^{-\beta \hat{H}})$ equals the functional integral over periodic paths in Euclidean time. This is a profound connection:
- **Trace**: sums over all eigenstates
- **Path integral**: sums over all closed paths in imaginary time
:::

:::{admonition} Derivation: Partition Function from Periodic Path Integral
:class: dropdown information

We show that $Z(\beta) = \text{Tr}(\mathrm{e}^{-\beta\hat{H}}) = \int_{\text{periodic}} \mathrm{e}^{-S_E[x]/\hbar}\,\mathcal{D}[x]$, where the functional integral runs over paths with $x(0) = x(\beta\hbar)$.

**Step 1: Write the trace as a position integral.**
Using position eigenstates $\{\vert x\rangle\}$ as a basis:

$$
Z(\beta) = \text{Tr}(\mathrm{e}^{-\beta\hat{H}}) = \int \mathrm{d}x\, \langle x|\mathrm{e}^{-\beta\hat{H}}|x\rangle = \int \mathrm{d}x\, K_E(x, \beta\hbar; x, 0)
$$

where $K_E(x', \tau; x, 0) = \langle x'|\mathrm{e}^{-\tau\hat{H}/\hbar}|x\rangle$ is the Euclidean propagator. The trace forces initial and final positions to coincide: $x(0) = x(\beta\hbar)$ — this is the origin of the **periodic boundary condition**.

**Step 2: Time-slice the Euclidean propagator.**
Divide the imaginary-time interval $[0, \beta\hbar]$ into $N$ steps of size $\epsilon = \beta\hbar/N$:

$$
\mathrm{e}^{-\beta\hat{H}} = \left(\mathrm{e}^{-\epsilon\hat{H}/\hbar}\right)^N
$$

Insert $N-1$ complete sets of position eigenstates $\int \mathrm{d}x_j\,\vert x_j\rangle\langle x_j\vert = \hat{I}$ between each factor:

$$
\langle x_N|\mathrm{e}^{-\beta\hat{H}}|x_0\rangle = \int \prod_{j=1}^{N-1}\mathrm{d}x_j \prod_{j=0}^{N-1} \langle x_{j+1}|\mathrm{e}^{-\epsilon\hat{H}/\hbar}|x_j\rangle
$$

**Step 3: Evaluate the short-imaginary-time matrix element.**
For small $\epsilon$, use $\hat{H} = \hat{p}^2/(2m) + V(\hat{x})$. Insert a complete set of momentum eigenstates and use the Trotter approximation $\mathrm{e}^{-\epsilon\hat{H}/\hbar} \approx \mathrm{e}^{-\epsilon\hat{p}^2/(2m\hbar)}\mathrm{e}^{-\epsilon V(\hat{x})/\hbar}$:

$$
\langle x_{j+1}|\mathrm{e}^{-\epsilon\hat{H}/\hbar}|x_j\rangle \approx \int \frac{\mathrm{d}p}{2\pi\hbar}\, \mathrm{e}^{\mathrm{i}p(x_{j+1}-x_j)/\hbar} \mathrm{e}^{-\epsilon p^2/(2m\hbar) - \epsilon V(x_j)/\hbar}
$$

Complete the square in $p$. The Gaussian integral $\int \frac{\mathrm{d}p}{2\pi\hbar} \mathrm{e}^{-\epsilon p^2/(2m\hbar) + \mathrm{i}p(x_{j+1}-x_j)/\hbar}$ evaluates to $\sqrt{\frac{m}{2\pi\hbar\epsilon}}\,\mathrm{e}^{-m(x_{j+1}-x_j)^2/(2\hbar\epsilon)}$. So:

$$
\langle x_{j+1}|\mathrm{e}^{-\epsilon\hat{H}/\hbar}|x_j\rangle \approx \sqrt{\frac{m}{2\pi\hbar\epsilon}}\exp\!\left(-\frac{m(x_{j+1}-x_j)^2}{2\hbar\epsilon} - \frac{\epsilon}{\hbar}V(x_j)\right)
$$

**Step 4: Assemble the full partition function.**
Set $x_N = x_0 = x$ (periodic BC) and integrate over $x$. Each of the $N$ short-time factors contributes. Combining all terms:

$$
Z(\beta) = \int \prod_{j=0}^{N-1}\mathrm{d}x_j\,\left(\frac{m}{2\pi\hbar\epsilon}\right)^{N/2} \exp\!\left(-\frac{1}{\hbar}\sum_{j=0}^{N-1}\left[\frac{m(x_{j+1}-x_j)^2}{2\epsilon} + \epsilon V(x_j)\right]\right)
$$

**Step 5: Take the continuum limit $N\to\infty$.**
Recognize that $\frac{(x_{j+1}-x_j)}{\epsilon} \to \dot{x}(\tau)$ and the sum becomes a Riemann integral:

$$
\frac{1}{\hbar}\sum_{j=0}^{N-1}\left[\frac{m(x_{j+1}-x_j)^2}{2\epsilon} + \epsilon V(x_j)\right] \xrightarrow{N\to\infty} \frac{1}{\hbar}\int_0^{\beta\hbar}\!\left[\frac{m}{2}\dot{x}^2 + V(x)\right]\mathrm{d}\tau = \frac{S_E[x]}{\hbar}
$$

The prefactor and integration measure define the functional measure $\mathcal{D}[x]$. The result is:

$$
Z(\beta) = \int_{x(0)=x(\beta\hbar)}\!\mathcal{D}[x]\,\mathrm{e}^{-S_E[x]/\hbar}
$$

**Physical interpretation.** The trace forces periodicity — we sum over all closed paths in imaginary time of period $\beta\hbar$. This is identical in structure to the real-time path integral (§3.2.1), but with $t \to -\mathrm{i}\tau$ and the paths forming closed loops. The classical saddle point $\delta S_E = 0$ among periodic paths gives the dominant contribution; quantum fluctuations around it are the corrections.
:::


### Summary

- **Wick rotation**: $t \to -\mathrm{i}\tau$ converts oscillating phases $\mathrm{e}^{\mathrm{i}S/\hbar}$ to decaying exponentials $\mathrm{e}^{-S_E/\hbar}$
- **Euclidean action**: $S_E = \int(\frac{m}{2}\dot{x}^2 + V)\,\mathrm{d}\tau$ — same form as Hamiltonian but in imaginary time
- **Partition function**: $Z(\beta) = \int_{\text{periodic}}\mathrm{e}^{-S_E/\hbar}\mathcal{D}[x]$ — path integral over periodic paths of period $\beta\hbar$
- **Temperature duality**: quantum mechanics in imaginary time $\beta\hbar$ is thermal statistical mechanics at temperature $T = 1/(k_B \beta)$


:::{admonition} Derivation: Matsubara Frequencies from Periodic Boundary Conditions
:class: dropdown information

The imaginary-time path integral $Z = \int_{\text{periodic}} \mathrm{e}^{-S_E/\hbar}\mathcal{D}[x]$ sums over paths satisfying $x(\beta\hbar) = x(0)$. Expanding in Fourier modes reveals a discrete frequency structure.

**Step 1: Fourier decomposition.** Any periodic path $x(\tau)$ on the interval $[0, \beta\hbar]$ with period $\beta\hbar$ can be expanded as

$$
x(\tau) = \frac{1}{\beta\hbar}\sum_{n=-\infty}^{\infty} \tilde{x}_n\, \mathrm{e}^{\mathrm{i}\omega_n \tau}, \qquad \omega_n = \frac{2\pi n}{\beta\hbar},
$$

where the discrete frequencies $\omega_n$ are the **bosonic Matsubara frequencies**. Periodicity $x(\beta\hbar) = x(0)$ is automatically satisfied since $\mathrm{e}^{\mathrm{i}\omega_n \beta\hbar} = \mathrm{e}^{2\pi\mathrm{i}n} = 1$.

**Step 2: Euclidean action in Fourier space.** For the harmonic oscillator $S_E = \int_0^{\beta\hbar} \left(\frac{m}{2}\dot{x}^2 + \frac{m\omega^2}{2}x^2\right)\mathrm{d}\tau$, substituting the Fourier expansion and using the Parseval identity:

$$
S_E = \frac{m}{2\beta\hbar}\sum_n (\omega_n^2 + \omega^2)\,|\tilde{x}_n|^2.
$$

Each Fourier mode decouples — the path integral becomes a **product of independent Gaussian integrals**, one per mode.

**Step 3: Partition function as a product.** Integrating over each $\tilde{x}_n$:

$$
Z = \prod_{n=-\infty}^{\infty} \sqrt{\frac{2\pi\hbar}{m(\omega_n^2 + \omega^2)/\beta\hbar}} = \prod_{n=-\infty}^{\infty} \sqrt{\frac{2\pi\beta\hbar^2}{m(\omega_n^2 + \omega^2)}}.
$$

**Step 4: Product formula for $\sinh$.** The Weierstrass product formula gives

$$
\prod_{n=1}^{\infty}\left(1 + \frac{\omega^2\beta^2\hbar^2}{4\pi^2 n^2}\right) = \frac{\sinh(\beta\hbar\omega/2)}{\beta\hbar\omega/2},
$$

and collecting the $n = 0$ mode and all positive and negative $n$:

$$
Z = \frac{1}{2\sinh(\beta\hbar\omega/2)},
$$

recovering the exact harmonic oscillator partition function. The Matsubara decomposition converts a functional integral into an infinite product of ordinary Gaussian integrals — a technique that extends to quantum field theory at finite temperature.
:::
