## Lecture Notes

### Overview

This section solves the Schrödinger equation for a free particle ($V = 0$) and builds the propagator from plane wave solutions. By superposing plane waves over all momenta and evaluating a Gaussian integral in momentum space, we recover the exact propagator:

$$K_\text{free}(x, t; x_0, 0) = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\frac{\mathrm{i}m}{2\hbar t}(x - x_0)^2\right)$$

The exponent contains the classical action for motion from $x_0$ to $x$ in time $t$. This confirms the central theme: **Phase = Action** is not metaphor—it is exact for the free particle.

### Plane Wave Solution

The free Schrödinger equation (with $V = 0$) is:

$$\mathrm{i}\hbar \frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} \psi$$

A natural ansatz is a **plane wave**:

$$\psi(x, t) = \mathrm{e}^{\mathrm{i}(kx - \omega t)}$$

:::{admonition} Substitute into the Schrödinger equation
:class: dropdown information

$$\mathrm{i}\hbar (-\mathrm{i}\omega) \mathrm{e}^{\mathrm{i}(kx - \omega t)} = -\frac{\hbar^2}{2m} (-k^2) \mathrm{e}^{\mathrm{i}(kx - \omega t)}$$

$$\hbar\omega = \frac{\hbar^2 k^2}{2m}$$

:::

:::{admonition} Plane Wave Dispersion Relation
:class: important

$$\omega(k) = \frac{\hbar k^2}{2m}$$

A plane wave with wave vector $k$ oscillates as $\exp(\mathrm{i}(kx - \omega t))$ and carries energy $E = \hbar\omega = \frac{\hbar^2 k^2}{2m} = \frac{p^2}{2m}$ (momentum eigenstate with $p = \hbar k$).
:::

### Building the Propagator from Plane Waves

The key insight: the propagator is a superposition of all plane wave solutions. A particle can take any momentum, so we sum over all $k$:

$$K_\text{free}(x, t; x_0, 0) = \int \frac{\mathrm{d}k}{(2\pi)^d} \, \mathrm{e}^{\mathrm{i}(k(x - x_0) - \omega(k) t)}$$

$$= \int \frac{\mathrm{d}k}{(2\pi)^d} \exp\left(\mathrm{i}k(x - x_0) - \mathrm{i}\frac{\hbar k^2}{2m} t\right)$$

This is exactly the form anticipated by the path integral: the phase is $S/\hbar$ where $S$ is the action, here summed over all possible momenta.

:::{admonition} Note: Plane Waves as Momentum Eigenstates
:class: note

Each plane wave with momentum $p = \hbar k$ is a momentum eigenstate $\vert p \rangle = (2\pi\hbar)^{-d/2} \mathrm{e}^{\mathrm{i}px/\hbar}$. The propagator is the momentum-space integral of the projector $\vert p \rangle \langle p \vert$ evolved by the time operator $\mathrm{e}^{-\mathrm{i}\hat{H}t/\hbar}$ with $\hat{H} = \hat{p}^2/(2m)$. This connection to spectral decomposition will be formalized later.
:::

### Evaluating the Gaussian Integral in Momentum Space

The integral over $k$ is Gaussian; completing the square and shifting variables yields $K_\text{free}$ in closed form.

:::{admonition} The Free-Particle Propagator
:class: important

$$K_\text{free}(x, t; x_0, 0) = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right)$$

This agrees with the one-step free kernel {eq}`eq-K-free-delta-t-one-step` from §3.2.2: set the elapsed time to $\delta t$ (so $t \to \delta t$) and identify endpoints as in time-slicing ($x_0 \leftrightarrow x'$). The Gaussian prefactor and phase then match exactly—**self-consistency** between the short-time path-integral factor and the exact momentum superposition.

:::

:::{admonition} Derivation: Gaussian Evaluation
:class: dropdown information

The exponent is quadratic in $k$:

$$
\begin{split}
\mathrm{i}k(x - x_0) - \mathrm{i}\frac{\hbar k^2}{2m} t
&= -\mathrm{i}\frac{\hbar t}{2m}\!\left(k^2 - \frac{2m}{\hbar t}k(x - x_0)\right)
\end{split}
$$

Complete the square in $k$:

$$
\begin{split}
k^2 - \frac{2m}{\hbar t}k(x - x_0)
&= \left(k - \frac{m}{\hbar t}(x - x_0)\right)^2 - \frac{m^2}{(\hbar t)^2}(x - x_0)^2
\end{split}
$$

The $k$ integral becomes

$$
\begin{split}
K_\text{free} = \int \frac{\mathrm{d}k}{(2\pi)^d} \exp\biggl(
&-\mathrm{i}\frac{\hbar t}{2m}\biggl[\left(k - \frac{m}{\hbar t}(x - x_0)\right)^2 \\
&\quad - \frac{m^2}{(\hbar t)^2}(x - x_0)^2\biggr]\biggr)
\end{split}
$$

The integral over the shifted quadratic term is a standard Gaussian:

$$
\begin{split}
\int \mathrm{d}k \exp\!\left(-\mathrm{i}\frac{\hbar t}{2m} k'^2\right)
&= \left(\frac{\pi m}{-\mathrm{i}\hbar t/2}\right)^{1/2} = \left(\frac{2\pi m}{\mathrm{i}\hbar t}\right)^{1/2}
\end{split}
$$

(where $k' = k - m(x - x_0)/(\hbar t)$ and we use $\sqrt{-\mathrm{i}} = (1-\mathrm{i})/\sqrt{2}$).

The remaining exponent is

$$
\begin{split}
\exp\!\left(\mathrm{i}\frac{\hbar t}{2m} \cdot \frac{m^2}{(\hbar t)^2}(x - x_0)^2\right)
&= \exp\!\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right)
\end{split}
$$

Combining,

$$
\begin{split}
K_\text{free}
&= \frac{1}{(2\pi)^d} \left(\frac{2\pi m}{\mathrm{i}\hbar t}\right)^{1/2} \exp\!\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right) \\
&= \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\!\left(\mathrm{i}\frac{m}{2\hbar t}(x - x_0)^2\right)
\end{split}
$$

:::



### Phase = Classical Action

The exponent can be rewritten by recognizing the classical action. A particle traveling from $x_0$ to $x$ in time $t$ with constant velocity has action:

$$
\begin{split}
S_\text{cl} = \int_0^t \frac{m}{2}\dot{x}^2 \, \mathrm{d}t'
= \frac{m}{2} v^2 t = \frac{m}{2t}(x - x_0)^2
\end{split}
$$

Thus the exponent in the propagator is precisely $\mathrm{i}S_\text{cl}/\hbar$:

$$K_\text{free} = \left(\frac{m}{2\pi \mathrm{i}\hbar t}\right)^{1/2} \exp\left(\mathrm{i}\frac{S_\text{cl}}{\hbar}\right)$$

:::{admonition} Phase = Action for the Free Particle
:class: attention

For the free particle, the propagator takes the form $K = A(t) \cdot \exp(\mathrm{i}S_\text{cl}/\hbar)$: the phase is exactly the classical action, and the prefactor $A(t) = (m/2\pi\mathrm{i}\hbar t)^{1/2}$ depends only on the elapsed time.

This remarkable simplification occurs because the free-particle action is quadratic in the path, so the path integral is an exact Gaussian. The ansatz $\psi \sim \exp(\mathrm{i}S/\hbar)$ from §3.1.3 becomes mathematically exact.
:::

### Closing the Circle

Recall the structure of unit 3.2:

- **§3.1**: Proposed $\psi \sim \mathrm{e}^{\mathrm{i}S/\hbar}$ as the guiding principle
- **§3.2.1**: Formalized the path integral as $\int \mathcal{D}x\,\mathrm{e}^{\mathrm{i}S/\hbar}$
- **§3.2.2**: Derived the Schrödinger equation from the path integral (initial value problem)
- **§3.2.3** (now): Solved the Schrödinger equation for the free particle and found $K \propto \mathrm{e}^{\mathrm{i}S_{\mathrm{cl}}/\hbar}$

The free particle closes the loop: the path integral — a sum over all paths weighted by $\mathrm{e}^{\mathrm{i}S[\text{path}]/\hbar}$ — yields a propagator whose phase is exactly the classical action. This happens because the free-particle action is quadratic, so the Gaussian integral is exact and only the classical path's action survives in the exponent.

For general potentials the action is not quadratic and the path integral cannot be evaluated in closed form. In §3.3, the stationary phase approximation will show that the classical path still dominates in the semiclassical limit, with $\mathrm{e}^{\mathrm{i}S_\text{cl}/\hbar}$ as the leading phase factor.

:::{admonition} Discussion: Why Is the Factor i Important?
:class: dropdown tip

The exponent is $+\mathrm{i}S_\text{cl}/\hbar$, not $-\mathrm{i}S_\text{cl}/\hbar$ (as in the path integral formulation itself, which has $\mathcal{S}/\hbar$ with $\mathcal{S}$ defined with a +i sign in the path integral definition). This is crucial:

- A particle with momentum $p = m v$ travels distance $\Delta x = v t = (p/m) t$, so $S_\text{cl} = p \Delta x / 2 = m (\Delta x)^2 / (2t) > 0$.
- With exponent $\mathrm{i}S_\text{cl}/\hbar > 0$, the phase oscillates: $\exp(\mathrm{i}\theta)$ with $\theta = m(\Delta x)^2 / (2\hbar t) \gg 1$ for macroscopic systems. This oscillation suppresses interference from distant paths (destructive interference).
- If the sign were reversed, far paths would constructively interfere—wrong physics.

The +i is the quantum signature of stationary-phase dominance.
:::

### Summary

- **Free-particle plane waves**: $\psi \sim \mathrm{e}^{\mathrm{i}(kx-\omega t)}$ with $\omega = \hbar k^2/(2m)$ solves the Schrödinger equation; each is a momentum eigenstate.
- **Propagator from momentum superposition**: Sum plane waves over all momenta (the path integral idea: paths have different momenta). Completing the square in momentum space gives a Gaussian in position.
- **Gaussian propagator**: $(m/2\pi \mathrm{i}\hbar t)^{1/2}\exp(\mathrm{i}S_{\mathrm{cl}}/\hbar)$ where $S_{\mathrm{cl}} = m(x-x')^2/(2t)$ is the classical action for constant-velocity motion.
- **Phase = Action for quadratic actions**: The exponent is exactly $\mathrm{i}S_{\mathrm{cl}}/\hbar$ because the free-particle action is quadratic, making the path integral an exact Gaussian. For general potentials, this becomes the leading-order semiclassical contribution (§3.3).
