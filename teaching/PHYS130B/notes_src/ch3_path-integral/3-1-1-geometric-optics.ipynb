## Lecture Notes

### Overview

The nature of light was one of the central debates in the history of physics. **Newton (1704)** proposed the *corpuscle theory*: light consists of particles flying in straight lines. **Huygens (1690)** proposed the *wave theory*: light is a wave propagating through a medium. Both theories correctly derive the three geometric laws of optics, but they make different predictions for the deeper nature of light and fail for different phenomena.

This lesson develops **geometric optics** from the particle picture. The central result is **Fermat's principle of stationary optical path**: a single variational principle from which all three laws of geometric optics follow—in the same way that Hamilton's principle governs classical mechanics.

### The Particle Theory of Light

In Newton's corpuscle model, light is a stream of particles. In a medium with **refractive index** $n$, the speed of light is

$$
v = \frac{c}{n}
$$ (eq-speed-of-light-medium)

where $c$ is the vacuum speed of light ($n = 1$ in vacuum, $n \approx 1.5$ in glass, $n \approx 1.33$ in water).

From the particle picture, three empirical laws describe how light rays behave:

| Law | Statement |
|-----|-----------|
| **Rectilinear propagation** | In a uniform medium, light travels in straight lines. |
| **Reflection** | At a smooth mirror, angle of incidence equals angle of reflection. |
| **Refraction** | At an interface between media $n_1$ and $n_2$: $n_1 \sin\theta_1 = n_2 \sin\theta_2$ (Snell's law). |

These laws describe mirrors, lenses, prisms, and optical fibers. All three follow from the variational principle below. The particle picture is powerful for geometric optics—but it provides no mechanism for **diffraction** (light bending around obstacles) or **interference** (two beams producing dark fringes). These phenomena prove decisive for the wave theory (§3.1.2).

### Fermat's Principle

:::{admonition} Fermat's Principle
:class: important

Light travels along the path for which the **optical path length** is stationary with respect to small variations:

$$
\delta L = 0, \quad L = \int_{\text{path}} n(\boldsymbol{r}) \, \mathrm{d}s
$$ (eq-fermat-principle)

Here, $n(\boldsymbol{r})$ is the refractive index and $\mathrm{d}s$ is the arc length element along the path.
:::

**Why "stationary" and not "minimum"?** In most cases the optical path length is a *minimum* for the true ray (Fermat's least-time principle). However, at caustics and some curved interfaces, the path can be a *maximum* or *saddle point*. The key is that the **first-order variation vanishes**: $\delta L = 0$.

**Connection to mechanics:** The optical path length $L = \int n \, \mathrm{d}s$ is the *action* for light, just as $S = \int L_{\rm mech} \, \mathrm{d}t$ is the mechanical action. Fermat's principle ($\delta L = 0$) is the exact analog of Hamilton's principle ($\delta S = 0$). This structural parallel becomes the bridge to quantum mechanics in §3.1.3.

### Laws of Geometric Optics

All three laws in the table above follow from Fermat's principle.

#### Rectilinear Propagation

In a uniform medium with constant $n$, the path that makes $L = \int n \, \mathrm{d}s$ stationary is simply a straight line.

#### Reflection

Reflection geometry showing $\theta_r=\theta_i$ measured from the normal.

:::

When light reflects off a smooth boundary, the law of reflection follows:

$$
\theta_{r} = \theta_{i}
$$

:::{admonition} Derivation: Reflection from Fermat's Principle
:class: dropdown information

Consider a ray traveling from source $A = (0, h_1)$ to target $B = (d, h_2)$, reflecting off a flat mirror at $y = 0$. The reflection point is $(x, 0)$.

The total path length (with constant $n$) is

$$
\ell(x) = \sqrt{x^2 + h_1^2} + \sqrt{(d - x)^2 + h_2^2}
$$

**Calculus approach.** Setting $\mathrm{d}\ell/\mathrm{d}x = 0$:

$$
\frac{x}{\sqrt{x^2 + h_1^2}} - \frac{d - x}{\sqrt{(d - x)^2 + h_2^2}} = 0
$$

The first term is $\sin\theta_i$ (sine of incidence angle) and the second is $\sin\theta_r$ (sine of reflection angle), so $\sin\theta_i = \sin\theta_r$, giving $\theta_i = \theta_r$.

**Image method.** Alternatively, reflect $B$ to its mirror image $B' = (d, -h_2)$. Any path $A \to (x, 0) \to B$ has the same length as $A \to (x, 0) \to B'$. The shortest such path is the straight line $A \to B'$, which gives $\theta_i = \theta_r$ by vertical symmetry.

To verify the extremum is a minimum, compute $\mathrm{d}^2\ell/\mathrm{d}x^2 = h_1^2 (x^2 + h_1^2)^{-3/2} + h_2^2 ((d-x)^2 + h_2^2)^{-3/2} > 0$.
:::

#### Refraction

:::{margin}
:::{figure} ../assets/figs/3-1-1-refraction-ray.png
:width: 90%
:alt: Refraction geometry across an interface with refractive indices n1 and n2

:::
:::

**Snell's Law**: At an interface between media with refractive indices $n_1$ and $n_2$:

$$
n_1 \sin \theta_1 = n_2 \sin \theta_2
$$ (eq-snell-law)

:::{admonition} Derivation: Snell's Law from Fermat's Principle
:class: dropdown information

Place the interface along $y = 0$, with source $A = (0, h_1)$ in medium $n_1$ and target $B = (d, -h_2)$ in medium $n_2$. The ray crosses the interface at $(x, 0)$. The total optical path length is

$$
L(x) = n_1 \sqrt{x^2 + h_1^2} + n_2 \sqrt{(d-x)^2 + h_2^2}
$$

**Step 1.** Differentiate and set $\mathrm{d}L / \mathrm{d}x = 0$:

$$
\frac{\mathrm{d}L}{\mathrm{d}x} = n_1 \frac{x}{\sqrt{x^2 + h_1^2}} - n_2 \frac{d - x}{\sqrt{(d-x)^2 + h_2^2}} = 0
$$

**Step 2.** Identify the geometric meaning of each term. From the ray geometry, $\sin \theta_1 = x / \sqrt{x^2 + h_1^2}$ is the sine of the incidence angle (measured from the normal), and $\sin \theta_2 = (d-x) / \sqrt{(d-x)^2 + h_2^2}$ is the sine of the refraction angle. Substituting:

$$
n_1 \sin\theta_1 = n_2 \sin\theta_2
$$

**Step 3.** Verify the extremum is a minimum: $\mathrm{d}^2 L / \mathrm{d}x^2 = n_1 h_1^2 (x^2 + h_1^2)^{-3/2} + n_2 h_2^2 ((d-x)^2 + h_2^2)^{-3/2} > 0$, confirming the stationary point is a minimum of optical path length.

**Physical interpretation.** The quantity $n \sin\theta$ is conserved across the interface. This is the **transverse component of the optical momentum** $n \hat{\boldsymbol{e}}_{\text{ray}}$, analogous to conservation of transverse momentum in mechanics when the potential depends only on $y$.
:::

### The Ray Equation

Fermat's principle can be applied to media where $n(\boldsymbol{r})$ varies continuously (**gradient-index** or GRIN media). The resulting **ray equation** is the Euler-Lagrange equation for the optical path length functional.

:::{admonition} Ray Equation
:class: important

For a ray path $\boldsymbol{r}(s)$ parametrized by arc length $s$ in a medium with refractive index $n(\boldsymbol{r})$:

$$
\frac{\mathrm{d}}{\mathrm{d}s}\!\left(n \frac{\mathrm{d}\boldsymbol{r}}{\mathrm{d}s}\right) = \boldsymbol{\nabla} n
$$ (eq-ray-equation)

:::

This is Newton's second law for light: the "optical momentum" $n \, \mathrm{d}\boldsymbol{r}/\mathrm{d}s$ changes in response to the "force" $\boldsymbol{\nabla} n$. Light bends toward regions of higher refractive index, just as a particle accelerates toward lower potential energy.

:::{admonition} Derivation: Ray Equation from Fermat's Principle
:class: dropdown information

Parametrize the ray path by arc length $s$, so $|\mathrm{d}\boldsymbol{r}/\mathrm{d}s| = 1$. The optical path length functional is

$$
L[\boldsymbol{r}] = \int n(\boldsymbol{r}(s)) \, \mathrm{d}s
$$

This has the form of a Lagrangian mechanics problem with "Lagrangian" $\mathcal{L} = n(\boldsymbol{r}) \, |\dot{\boldsymbol{r}}|$ where $\dot{\boldsymbol{r}} = \mathrm{d}\boldsymbol{r}/\mathrm{d}s$. Since $|\dot{\boldsymbol{r}}| = 1$ on a path parametrized by arc length, $\mathcal{L} = n(\boldsymbol{r})$.

**Euler-Lagrange equation.** For the functional $L = \int \mathcal{L}(\boldsymbol{r}, \dot{\boldsymbol{r}}) \, \mathrm{d}s$ with the constraint $|\dot{\boldsymbol{r}}| = 1$, the Euler-Lagrange equation gives:

$$
\frac{\mathrm{d}}{\mathrm{d}s}\frac{\partial(n |\dot{\boldsymbol{r}}|)}{\partial \dot{\boldsymbol{r}}} = \frac{\partial(n |\dot{\boldsymbol{r}}|)}{\partial \boldsymbol{r}}
$$

Computing each side:

$$
\frac{\partial(n |\dot{\boldsymbol{r}}|)}{\partial \dot{\boldsymbol{r}}} = n \frac{\dot{\boldsymbol{r}}}{|\dot{\boldsymbol{r}}|} = n \dot{\boldsymbol{r}}, \qquad \frac{\partial(n |\dot{\boldsymbol{r}}|)}{\partial \boldsymbol{r}} = |\dot{\boldsymbol{r}}| \, \boldsymbol{\nabla} n = \boldsymbol{\nabla} n
$$

Substituting yields the ray equation:

$$
\frac{\mathrm{d}}{\mathrm{d}s}\!\left(n \frac{\mathrm{d}\boldsymbol{r}}{\mathrm{d}s}\right) = \boldsymbol{\nabla} n
$$

**Recovery of Snell's law.** At a sharp interface (step-function $n$), integrate the ray equation across the boundary. The component of $n \, \mathrm{d}\boldsymbol{r}/\mathrm{d}s$ tangent to the interface is continuous, giving $n_1 \sin\theta_1 = n_2 \sin\theta_2$.

**Analogy with Newton's second law.** Writing $\boldsymbol{p}_{\text{opt}} = n \hat{\boldsymbol{e}}$ where $\hat{\boldsymbol{e}} = \mathrm{d}\boldsymbol{r}/\mathrm{d}s$ is the ray direction, the ray equation reads $\mathrm{d}\boldsymbol{p}_{\text{opt}}/\mathrm{d}s = \boldsymbol{\nabla} n$. Light bends toward regions of higher $n$, just as a particle accelerates toward lower potential.
:::

:::{admonition} Derivation: The Eikonal Equation from the Wave Equation
:class: dropdown information

The eikonal equation is the mathematical bridge between wave optics and geometric optics. We derive it by inserting a WKB-like ansatz into the scalar wave equation and taking the short-wavelength limit.

**Setup.** In a medium with spatially varying refractive index $n(\boldsymbol{r})$, a monochromatic wave at frequency $\omega$ satisfies the **Helmholtz equation**:

$$
\nabla^2 u + k_0^2 n^2(\boldsymbol{r}) u = 0, \qquad k_0 = \frac{\omega}{c}
$$

**Step 1: WKB ansatz.** Write the wave field as

$$
u(\boldsymbol{r}) = A(\boldsymbol{r})\, \mathrm{e}^{\mathrm{i} k_0 S(\boldsymbol{r})}
$$

where $A(\boldsymbol{r})$ is a slowly varying real amplitude and $S(\boldsymbol{r})$ is the **eikonal** (optical phase function, measured in units of length). Surfaces of constant $S$ are the wavefronts.

**Step 2: Compute the Laplacian.** Applying $\nabla$ twice to the ansatz:

$$
\nabla u = \left(\nabla A + \mathrm{i} k_0 A\, \nabla S\right)\mathrm{e}^{\mathrm{i} k_0 S}
$$

$$
\nabla^2 u = \left[\nabla^2 A + 2\mathrm{i} k_0 (\nabla A)\cdot(\nabla S) + \mathrm{i} k_0 A\, \nabla^2 S - k_0^2 A\, |\nabla S|^2\right]\mathrm{e}^{\mathrm{i} k_0 S}
$$

**Step 3: Substitute into the Helmholtz equation.** Dividing by $A\,\mathrm{e}^{\mathrm{i} k_0 S}$:

$$
\frac{\nabla^2 A}{A} + 2\mathrm{i} k_0 \frac{\nabla A}{A}\cdot\nabla S + \mathrm{i} k_0 \nabla^2 S - k_0^2 |\nabla S|^2 + k_0^2 n^2 = 0
$$

**Step 4: Separate real and imaginary parts.**

$$
\underbrace{\left(n^2 - |\nabla S|^2\right)k_0^2 + \frac{\nabla^2 A}{A}}_{\text{real}} + \mathrm{i}\underbrace{\left(2\nabla A\cdot\nabla S + A\,\nabla^2 S\right)k_0}_{\text{imaginary}} = 0
$$

**Step 5: Short-wavelength limit.** In the geometric optics regime $k_0 \to \infty$ (i.e., wavelength $\lambda = 2\pi/k_0 \to 0$), the term $\nabla^2 A/A$ is $O(1)$ and becomes negligible compared to the $O(k_0^2)$ terms. Setting each part to zero at leading order:

- **Real part** ($O(k_0^2)$): the **Eikonal equation**

$$
|\nabla S|^2 = n^2(\boldsymbol{r})
$$ (eq-eikonal)

- **Imaginary part** ($O(k_0)$): the **Transport equation**

$$
2\nabla A\cdot\nabla S + A\,\nabla^2 S = 0 \quad \Longleftrightarrow \quad \nabla\cdot\left(A^2\,\nabla S\right) = 0
$$

The transport equation says that the intensity $I = A^2$ is conserved along ray tubes.

**Step 6: Ray direction from the eikonal.** The eikonal $S$ plays the role of Hamilton's principal function. Comparing with Fermat's principle, the optical momentum is

$$
\boldsymbol{p}_{\text{opt}} = \nabla S, \qquad |\boldsymbol{p}_{\text{opt}}| = n
$$

so $\nabla S$ is everywhere tangent to the rays with magnitude $n$. The ray direction is $\hat{\boldsymbol{e}} = \nabla S / n$.

**Step 7: Recover the ray equation.** Differentiate the eikonal equation along the ray (parametrized by arc length $s$):

$$
\frac{\mathrm{d}}{\mathrm{d}s}(n\hat{\boldsymbol{e}}) = \frac{\mathrm{d}}{\mathrm{d}s}(\nabla S) = (\hat{\boldsymbol{e}}\cdot\nabla)\nabla S
$$

Using $\hat{\boldsymbol{e}} = \nabla S/n$ and $|\nabla S|^2 = n^2$:

$$
(\hat{\boldsymbol{e}}\cdot\nabla)\nabla S = \frac{(\nabla S\cdot\nabla)\nabla S}{n} = \frac{\nabla|\nabla S|^2/2}{n} = \frac{\nabla n^2/2}{n} = \nabla n
$$

This reproduces the **ray equation** $\mathrm{d}(n\hat{\boldsymbol{e}})/\mathrm{d}s = \nabla n$, confirming consistency.

**Analogy with Hamilton-Jacobi mechanics.** The eikonal equation $|\nabla S|^2 = n^2$ is structurally identical to the classical Hamilton-Jacobi equation $|\nabla S_{\text{mech}}|^2/2m + V = E$ for a particle with energy $E$ in potential $V$. The identification $n^2 \leftrightarrow 2m(E - V)$ is the optical-mechanical analogy that Schrödinger used to develop wave mechanics — and that the path integral makes precise.
:::


:::{admonition} Derivation: Snell's Law as Conservation of Canonical Momentum
:class: dropdown information

In classical mechanics, a cyclic coordinate (one that does not appear in the Lagrangian) has a conserved conjugate momentum. The same principle governs ray optics.

**Setup.** Parameterize a ray in a 2D plane by the horizontal coordinate $x$, with $y(x)$ describing the ray path. The **optical Lagrangian** is the optical path length per unit $x$:

$$
\mathcal{L}(y, y') = n(x, y)\sqrt{1 + y^{\prime 2}}, \qquad y' = \frac{\mathrm{d}y}{\mathrm{d}x}.
$$

**Canonical momentum.** The canonical momentum conjugate to $y$ is

$$
p = \frac{\partial \mathcal{L}}{\partial y'} = \frac{n\, y'}{\sqrt{1 + y^{\prime 2}}} = n\sin\theta,
$$

where $\theta$ is the angle the ray makes with the $x$-axis. This is the **transverse momentum** of the ray.

**Euler-Lagrange equation.** The equation of motion is

$$
\frac{\mathrm{d}}{\mathrm{d}x}\left(n\sin\theta\right) = \frac{\partial \mathcal{L}}{\partial y} = \frac{\partial n}{\partial y}\sqrt{1+y^{\prime 2}}.
$$

**Conservation and Snell's law.** If the medium is **stratified** — $n = n(x)$ independent of $y$ — then $\partial \mathcal{L}/\partial y = 0$ and the canonical momentum is **conserved**:

$$
p = n(x)\sin\theta(x) = \text{const}.
$$

At a sharp interface between media $n_1$ and $n_2$, continuity of $p$ gives

$$
n_1 \sin\theta_1 = n_2 \sin\theta_2.
$$

This is Snell's law — not an independent postulate, but a **conservation law** following from the translational symmetry of the medium. The analogy with mechanics is perfect: in a uniform potential, the component of momentum along the symmetry direction is conserved.

**Paraxial limit.** For small angles ($\sin\theta \approx \theta$), $p \approx n\theta$. The canonical mechanics description becomes linear — the **ABCD matrix formalism** for paraxial optics — in exact parallel with the linearization of Hamilton's equations near equilibrium.
:::

### Summary

- **Historical debate**: Newton's corpuscle theory and Huygens' wave theory both predict the same geometric laws, but the corpuscle picture offers no mechanism for diffraction or interference. These phenomena ultimately reveal the wave nature of light (§3.1.2).

- **Speed of light in media**: $v = c/n$; the refractive index $n > 1$ encodes how a medium slows light.

- **Fermat's Principle**: Light travels paths where the optical path length $L = \int n \, \mathrm{d}s$ is stationary. All three laws of geometric optics follow from $\delta L = 0$.

- **Ray equation**: In a continuously varying medium, $\frac{\mathrm{d}}{\mathrm{d}s}(n \, \mathrm{d}\boldsymbol{r}/\mathrm{d}s) = \boldsymbol{\nabla} n$. Light bends toward regions of higher refractive index.

- **Action parallel**: Optical path length plays the role of action for light; Fermat's principle ($\delta L = 0$) mirrors Hamilton's principle ($\delta S = 0$). This parallel is the key to the path integral (§3.1.3).
