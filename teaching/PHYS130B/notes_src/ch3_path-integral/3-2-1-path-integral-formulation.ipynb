## Lecture Notes

### Overview

Section 3.1.3 ended with the core idea: quantization means **action becomes the phase exponent**, and quantum mechanics is a **sum over all paths**. This section makes that precise. We introduce the **propagator** — the fundamental transition amplitude — express it as Feynman's **path integral** (sum over paths weighted by $\mathrm{e}^{\mathrm{i}S/\hbar}$), and decode the functional integral via **time-slicing** into concrete integrals over single time steps. The result: a complete, equivalent reformulation of quantum mechanics that makes the classical limit transparent.

### The Propagator: Transition Amplitude

:::{admonition} Propagator Definition
:class: important

The **propagator** $K(x, t; x', t')$ is the quantum amplitude for a particle at position $x'$ at time $t'$ to be found at position $x$ at time $t > t'$:

$$
K(x, t; x', t') = \langle x \vert \hat{U}(t, t') \vert x' \rangle
$$

where $\hat{U}(t, t') = \mathrm{e}^{-\mathrm{i}\hat{H}(t-t')/\hbar}$ is the time-evolution operator.
:::

- It determines everything: Given any initial state $\psi(x', t')$, the state at any later time is

    $$
    \psi(x, t) = \int K(x, t; x', t') \, \psi(x', t') \, \mathrm{d}x'
    $$

:::{admonition} Derivation: Wavefunction Evolution from the Propagator
:class: dropdown information

Start from $\vert\psi(t)\rangle = \hat{U}(t,t')\vert\psi(t')\rangle$ and project onto position:

$$
\psi(x,t) = \langle x\vert\psi(t)\rangle = \langle x\vert \hat{U}(t,t')\vert\psi(t')\rangle
$$

Insert the resolution of identity $\hat{I} = \int \vert x'\rangle\langle x'\vert\,\mathrm{d}x'$ between $\hat{U}$ and $\vert\psi(t')\rangle$:

$$
\psi(x,t) = \int \langle x\vert \hat{U}(t,t')\vert x'\rangle \langle x'\vert\psi(t')\rangle\,\mathrm{d}x' = \int K(x,t;x',t')\,\psi(x',t')\,\mathrm{d}x'
$$

where we identified $\langle x\vert \hat{U}(t,t')\vert x'\rangle = K(x,t;x',t')$ and $\langle x'\vert\psi(t')\rangle = \psi(x',t')$. The propagator acts as the integral kernel (Green's function) of the time-evolution operator in the position basis.
:::

- The propagator also satisfies a **composition property**: you can split a path at any intermediate time $t_m$:

    $$
    K(x, t; x', t') = \int K(x, t; x_m, t_m) \, K(x_m, t_m; x', t') \, \mathrm{d}x_m
    $$

    This follows from inserting the identity $\int \vert x_m \rangle \langle x_m \vert \, \mathrm{d}x_m$ into the product $\hat{U}(t, t') = \hat{U}(t, t_m) \hat{U}(t_m, t')$.

:::{admonition} Derivation: Composition Property
:class: dropdown information

The time-evolution operator satisfies the group property $\hat{U}(t,t') = \hat{U}(t,t_m)\hat{U}(t_m,t')$ for any $t' < t_m < t$. Writing the propagator in terms of $\hat{U}$:

$$
K(x,t;x',t') = \langle x\vert \hat{U}(t,t')\vert x'\rangle = \langle x\vert \hat{U}(t,t_m)\hat{U}(t_m,t')\vert x'\rangle
$$

Insert the resolution of identity $\hat{I} = \int \vert x_m\rangle\langle x_m\vert\,\mathrm{d}x_m$ between the two evolution operators:

$$
= \int \langle x\vert \hat{U}(t,t_m)\vert x_m\rangle\langle x_m\vert \hat{U}(t_m,t')\vert x'\rangle\,\mathrm{d}x_m
$$

Recognizing each matrix element as a propagator:

$$
= \int K(x,t;x_m,t_m)\,K(x_m,t_m;x',t')\,\mathrm{d}x_m
$$

This is the quantum analog of the Chapman–Kolmogorov equation in probability theory: to go from $x'$ to $x$, sum over all possible intermediate positions $x_m$ at time $t_m$. The property generalizes to any number of intermediate times by repeated insertion of the identity.
:::

### Feynman's Path Integral

:::{admonition} The Path Integral
:class: important

The propagator equals a **functional integral over all paths** from $(x', t')$ to $(x, t)$:

$$
K(x, t; x', t') = \int_{x(t')=x'}^{x(t)=x} \mathrm{e}^{\mathrm{i}S[x]/\hbar} \, \mathcal{D}[x]
$$

where $S[x] = \int_{t'}^{t} L(x, \dot{x}, \tau) \, \mathrm{d}\tau$ is the action along path $x(\tau)$.

:::

Each path contributes with **unit probability weight** $|\mathrm{e}^{\mathrm{i}S/\hbar}|=1$ but **different phase** $S/\hbar$. Paths with similar action interfere constructively; paths with wildly different action oscillate and cancel.

This is the **stationary-phase principle** from §3.1.3 realized exactly: the classical path (where $\delta S = 0$) is surrounded by neighboring paths with nearly equal phase, so they all add coherently. Off the classical path, the phase varies rapidly and neighboring paths cancel. Thus the classical path dominates, and quantum effects emerge from the sum of near-classical paths.

:::{admonition} Discussion: Is the Functional Integral Real?
:class: dropdown tip

Summing over uncountably many paths seems ill-defined. Yet the path integral must give the same results as the Schrödinger equation—both describe the same physics. The resolution is **time-slicing**: discretize time, perform finitely many ordinary integrals, then take the continuum limit carefully. The mathematical details are subtle, but the result is rigorous and equivalent to Schrödinger.
:::

### The Infinitesimal Propagator: One Time Slice

For a single small time step $\delta t$, we focus on the **short-time propagator** $K_{\delta t}(x, x')$ from $x'$ to $x$ in time $\delta t$.

For a nonrelativistic particle with Lagrangian $L = \tfrac{1}{2}m\dot{x}^2 - V(x)$, the action over one slice (midpoint rule for the potential) is

$$
S_{\text{slice}} = \tfrac{1}{2}m\frac{(x - x')^2}{\delta t} - V\!\left(\frac{x+x'}{2}\right) \delta t
$$

:::{admonition} One-step propagator (1D)
:class: important

The Gaussian integration over momentum yields the **short-time propagator** for one step $\delta t$:

$$
K_{\delta t}(x, x') = \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2} \exp\left\{\frac{\mathrm{i}}{\hbar}\left[\frac{m(x-x')^2}{2\,\delta t} - V\!\left(\frac{x+x'}{2}\right)\delta t\right]\right\}
$$ (eq-K-delta-t-one-step)

:::

For a **free particle** ($V=0$), this reduces to 

$$K_{\delta t}^{\text{free}}(x, x') = \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2} \exp\!\left(\frac{\mathrm{i}m(x-x')^2}{2\hbar\,\delta t}\right).$$


:::{admonition} Derivation: Short-Time Propagator via Momentum Integral
:class: dropdown information

**Step 1: Trotter-Suzuki splitting.** Over an infinitesimal time step, $\hat{U}(\delta t) = \mathrm{e}^{-\mathrm{i}\hat{H}\delta t/\hbar}$ with $\hat{H} = \tfrac{\hat{p}^2}{2m} + V(\hat{x})$. Since $\hat{p}$ and $\hat{x}$ do not commute, we use the symmetric Trotter-Suzuki decomposition (accurate to order $\delta t^2$):

$$
\mathrm{e}^{-\mathrm{i}\hat{H}\delta t/\hbar} \approx \mathrm{e}^{-\mathrm{i}V(\hat{x})\delta t/2\hbar} \, \mathrm{e}^{-\mathrm{i}\hat{p}^2\delta t/2m\hbar} \, \mathrm{e}^{-\mathrm{i}V(\hat{x})\delta t/2\hbar} + O(\delta t^3)
$$

**Step 2: Position matrix element.** Sandwich between $\langle x\vert$ and $\vert x'\rangle$. The potential operators act on position eigenstates: $\mathrm{e}^{-\mathrm{i}V(\hat{x})\delta t/2\hbar}\vert x'\rangle = \mathrm{e}^{-\mathrm{i}V(x')\delta t/2\hbar}\vert x'\rangle$ and $\langle x\vert\mathrm{e}^{-\mathrm{i}V(\hat{x})\delta t/2\hbar} = \langle x\vert\mathrm{e}^{-\mathrm{i}V(x)\delta t/2\hbar}$. This gives:

$$
K_{\delta t}(x,x') = \mathrm{e}^{-\mathrm{i}[V(x)+V(x')]\delta t/2\hbar}\,\langle x\vert\mathrm{e}^{-\mathrm{i}\hat{p}^2\delta t/2m\hbar}\vert x'\rangle
$$

Using the midpoint approximation $\tfrac{1}{2}[V(x)+V(x')] \approx V\!\bigl(\tfrac{x+x'}{2}\bigr) + O(\delta t^2)$, the potential phase becomes $\mathrm{e}^{-\mathrm{i}V(\frac{x+x'}{2})\delta t/\hbar}$.

**Step 3: Momentum resolution of identity.** Insert $\hat{I} = \int \vert p\rangle\langle p\vert\,\tfrac{\mathrm{d}p}{2\pi\hbar}$ into the kinetic piece:

$$
\langle x\vert\mathrm{e}^{-\mathrm{i}\hat{p}^2\delta t/2m\hbar}\vert x'\rangle = \int \frac{\mathrm{d}p}{2\pi\hbar}\,\langle x\vert p\rangle\,\mathrm{e}^{-\mathrm{i}p^2\delta t/2m\hbar}\,\langle p\vert x'\rangle
$$

Using $\langle x\vert p\rangle = \mathrm{e}^{\mathrm{i}px/\hbar}$ and $\langle p\vert x'\rangle = \mathrm{e}^{-\mathrm{i}px'/\hbar}$:

$$
= \int \frac{\mathrm{d}p}{2\pi\hbar}\,\exp\!\left[\frac{\mathrm{i}}{\hbar}\left(p(x-x') - \frac{p^2\delta t}{2m}\right)\right]
$$

**Step 4: Gaussian integral.** This is a standard Gaussian $\int \mathrm{e}^{-ap^2+bp}\,\mathrm{d}p = \sqrt{\pi/a}\,\mathrm{e}^{b^2/4a}$ with $a = \mathrm{i}\delta t/(2m\hbar)$ and $b = \mathrm{i}(x-x')/\hbar$. Completing the square in the exponent:

$$
-\frac{\mathrm{i}\delta t}{2m\hbar}p^2 + \frac{\mathrm{i}(x-x')}{\hbar}p = -\frac{\mathrm{i}\delta t}{2m\hbar}\left[p - \frac{m(x-x')}{\delta t}\right]^2 + \frac{\mathrm{i}m(x-x')^2}{2\hbar\,\delta t}
$$

The shifted Gaussian integral gives $\sqrt{2\pi m\hbar/(\mathrm{i}\delta t)}$, and dividing by $2\pi\hbar$:

$$
\langle x\vert\mathrm{e}^{-\mathrm{i}\hat{p}^2\delta t/2m\hbar}\vert x'\rangle = \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2}\exp\!\left(\frac{\mathrm{i}m(x-x')^2}{2\hbar\,\delta t}\right)
$$

**Step 5: Combine.** Multiplying by the potential phase factor:

$$
K_{\delta t}(x,x') = \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2}\exp\!\left\{\frac{\mathrm{i}}{\hbar}\left[\frac{m(x-x')^2}{2\,\delta t} - V\!\left(\frac{x+x'}{2}\right)\delta t\right]\right\}
$$

The exponent is exactly $\mathrm{i}S_{\text{slice}}/\hbar$ with the classical one-step action, confirming {eq}`eq-K-delta-t-one-step`. The prefactor $(m/2\pi\mathrm{i}\hbar\,\delta t)^{1/2}$ arises purely from the Gaussian integration and ensures the propagator is correctly normalized.
:::

### Time-Slicing: Turning Paths into Integrals

Divide the total time interval from $t'$ to $t$ into $N$ small steps of duration $\delta t = (t-t')/N$. Insert complete sets of position eigenstates at each intermediate time $t_0 < t_1 < \cdots < t_N$:

$$
\int \vert x_n \rangle \langle x_n \vert \, \mathrm{d}x_n = \hat{I}
$$

The full propagator factorizes:

$$
K(x, t; x', t') = \int \prod_{n=1}^{N-1} \mathrm{d}x_n \, \prod_{n=0}^{N-1} K_{\delta t}(x_{n+1}, x_n)
$$

Each factor $K_{\delta t}$ is the one-step kernel {eq}`eq-K-delta-t-one-step`. Multiplying the exponentials builds up the total phase along a discretized path; in the limit $N \to \infty$ ($\delta t \to 0$), this becomes the continuum path integral.

:::{admonition} Derivation: Path Integral Measure from Time-Slicing
:class: dropdown information

**Step 1: Expand each one-step propagator.** Substituting {eq}`eq-K-delta-t-one-step` into the factorized propagator, each slice contributes a prefactor and an exponential:

$$
K(x,t;x',t') = \int \prod_{n=1}^{N-1}\mathrm{d}x_n \prod_{n=0}^{N-1}\left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{1/2}\exp\!\left\{\frac{\mathrm{i}}{\hbar}\left[\frac{m(x_{n+1}-x_n)^2}{2\,\delta t} - V\!\left(\frac{x_{n+1}+x_n}{2}\right)\delta t\right]\right\}
$$

where $x_0 = x'$ and $x_N = x$ are the fixed endpoints.

**Step 2: Combine the exponentials.** Since $\mathrm{e}^{A}\mathrm{e}^{B} = \mathrm{e}^{A+B}$ for c-numbers, the product of exponentials becomes a single exponential of the sum:

$$
\prod_{n=0}^{N-1}\exp(\cdots) = \exp\!\left\{\frac{\mathrm{i}}{\hbar}\sum_{n=0}^{N-1}\left[\frac{m(x_{n+1}-x_n)^2}{2\,\delta t} - V\!\left(\frac{x_{n+1}+x_n}{2}\right)\delta t\right]\right\}
$$

The sum in the exponent is the **discretized action** $S_N[\{x_n\}]$ — a Riemann sum approximation to $S[x] = \int_{t'}^{t}[\tfrac{1}{2}m\dot{x}^2 - V(x)]\,\mathrm{d}\tau$, where $\dot{x} \approx (x_{n+1}-x_n)/\delta t$.

**Step 3: Collect prefactors.** There are $N$ one-step propagators, each contributing $(m/2\pi\mathrm{i}\hbar\,\delta t)^{1/2}$. Define the overall normalization:

$$
\mathcal{A}_N \equiv \left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{N/2}
$$

**Step 4: Define the path integral measure.** The full expression is:

$$
K(x,t;x',t') = \mathcal{A}_N \int \prod_{n=1}^{N-1}\mathrm{d}x_n\;\exp\!\left(\frac{\mathrm{i}}{\hbar}S_N[\{x_n\}]\right)
$$

In the continuum limit $N\to\infty$ ($\delta t\to 0$), we write this symbolically as:

$$
K(x,t;x',t') = \int_{x(t')=x'}^{x(t)=x}\mathcal{D}[x]\;\mathrm{e}^{\mathrm{i}S[x]/\hbar}
$$

where the **functional measure** absorbs both the intermediate integrations and the normalization:

$$
\mathcal{D}[x] \equiv \lim_{N\to\infty}\mathcal{A}_N\prod_{n=1}^{N-1}\mathrm{d}x_n = \lim_{N\to\infty}\left(\frac{m}{2\pi\mathrm{i}\hbar\,\delta t}\right)^{N/2}\prod_{n=1}^{N-1}\mathrm{d}x_n
$$

The measure $\mathcal{D}[x]$ is not an ordinary measure — its normalization diverges as $N\to\infty$, but the full expression (normalization times oscillatory integral) converges to the correct propagator. This is the precise mathematical content of Feynman's path integral.
:::

:::{admonition} Connection to Huygens' Principle
:class: note

Compare this to Huygens' principle (§3.1.2): each point $x_n$ at time $t_n$ acts as a source of matter wavelets that propagate to $x_{n+1}$, with phase proportional to the action $S_{\text{slice}}$. The time-sliced path integral is exactly this: **a chain of wavelets**, each with phase $\mathrm{e}^{\mathrm{i}S_{\text{slice}}/\hbar}$, summed over all paths.
:::


:::{admonition} Derivation: Classical Limit of the Path Integral
:class: dropdown information

In the classical limit $\hbar \to 0$, the phase $\mathrm{e}^{\mathrm{i}S[\boldsymbol{x}]/\hbar}$ oscillates wildly for most paths, causing destructive interference. Only paths near the **stationary phase point** — where $\delta S = 0$ — contribute significantly. This recovers classical mechanics from the path integral.

**Step 1: Functional variation.** Consider a small deviation $x(t) \to x(t) + \delta x(t)$ with fixed endpoints $\delta x(t_i) = \delta x(t_f) = 0$. The action

$$
S[x] = \int_{t_i}^{t_f} L(x, \dot{x})\,\mathrm{d}t = \int_{t_i}^{t_f} \left(\frac{1}{2}m\dot{x}^2 - V(x)\right)\mathrm{d}t
$$

changes by

$$
\delta S = \int_{t_i}^{t_f} \left(\frac{\partial L}{\partial x}\,\delta x + \frac{\partial L}{\partial \dot{x}}\,\delta\dot{x}\right)\mathrm{d}t
$$

**Step 2: Integration by parts.** The second term gives

$$
\int_{t_i}^{t_f} \frac{\partial L}{\partial \dot{x}}\,\delta\dot{x}\,\mathrm{d}t = \left[\frac{\partial L}{\partial \dot{x}}\,\delta x\right]_{t_i}^{t_f} - \int_{t_i}^{t_f} \frac{\mathrm{d}}{\mathrm{d}t}\frac{\partial L}{\partial \dot{x}}\,\delta x\,\mathrm{d}t
$$

The boundary term vanishes because $\delta x(t_i) = \delta x(t_f) = 0$.

**Step 3: Euler-Lagrange equation.** For $\delta S = 0$ to hold for *all* variations $\delta x(t)$, the integrand must vanish:

$$
\frac{\mathrm{d}}{\mathrm{d}t}\frac{\partial L}{\partial \dot{x}} = \frac{\partial L}{\partial x}
$$

For the Lagrangian $L = \frac{1}{2}m\dot{x}^2 - V(x)$, this gives Newton's second law $m\ddot{x} = -V'(x)$.

**Step 4: Semiclassical expansion.** Writing $x(t) = x_{\mathrm{cl}}(t) + \xi(t)$ where $x_{\mathrm{cl}}$ satisfies the classical EOM and $\xi$ is a fluctuation with $\xi(t_i) = \xi(t_f) = 0$:

$$
S[x_{\mathrm{cl}} + \xi] = S[x_{\mathrm{cl}}] + \underbrace{\delta S\big|_{x_{\mathrm{cl}}}}_{=\,0} + \frac{1}{2}\int_{t_i}^{t_f} \xi(t)\left(-m\frac{\mathrm{d}^2}{\mathrm{d}t^2} - V''(x_{\mathrm{cl}})\right)\xi(t)\,\mathrm{d}t + \cdots
$$

The linear term vanishes because $x_{\mathrm{cl}}$ is stationary. The path integral becomes

$$
K \approx \mathrm{e}^{\mathrm{i}S[x_{\mathrm{cl}}]/\hbar} \int \mathcal{D}[\xi]\,\exp\!\left(\frac{\mathrm{i}}{2\hbar}\int \xi\,\hat{O}\,\xi\,\mathrm{d}t\right)
$$

where $\hat{O} = -m\ddot{} - V''(x_{\mathrm{cl}})$ is the fluctuation operator. The Gaussian functional integral over $\xi$ gives the Van Vleck prefactor $(\det \hat{O})^{-1/2}$, which is the **semiclassical approximation** developed in §3.3.

**Summary.** The classical limit extracts the single path of stationary action; quantum corrections come from Gaussian fluctuations around it. The path integral thus contains classical mechanics as its $\hbar \to 0$ leading term.
:::
### Summary

- The **propagator** $K(x, t; x', t') = \langle x \vert \hat{U} \vert x' \rangle$ is the central transition amplitude; it encodes the full time evolution.
- **Feynman's path integral** expresses $K$ as a sum over all histories, each weighted by $\mathrm{e}^{\mathrm{i}S/\hbar}$. The classical path dominates via **stationary phase**; quantum paths contribute through interference.
- The **one-step propagator** $K_{\delta t}$ is given explicitly in {eq}`eq-K-delta-t-one-step` (free-particle limit and $d$-dimensional prefactor noted there).
- **Time-slicing** converts the functional integral into a sequence of ordinary integrals over intermediate positions, which converges to the continuum path integral as $N \to \infty$.
