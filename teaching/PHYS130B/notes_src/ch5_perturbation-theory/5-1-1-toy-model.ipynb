## Lecture Notes

Perturbation theory is the art of solving the unsolvable: when exact Hamiltonians resist solution, we leverage nearby solvable systems and expand the difference away. The qubit—a two-level system with an exact solution—becomes our laboratory for understanding how perturbative approximations work, how power series in a coupling constant reveal the structure of quantum states, and crucially, where and why those series break down.
### What is Perturbation Theory?

:::{important}
:class: important

Convergence condition: Perturbation theory requires $\lambda \|V\| \ll \min(\Delta E)$, where $\Delta E$ is the smallest unperturbed energy gap.
:::

:::{caution}
:class: cautio

n

Finite radius of convergence: Even for analytic exact solutions, the perturbative series may diverge beyond $|\lambda| = R$ due to singularities in the complex $\lambda$-plane.
:::

**Perturbation theory** is a systematic approximation method that extends our knowledge of exactly solvable quantum systems to nearby regimes in parameter space. Rather than solving a complex Hamiltonian from scratch, we leverage solutions of a simpler (solvable) system and treat the difference as a small perturbation.

**Central idea**: Expand eigenvalues and eigenstates as power series in a small coupling parameter $\lambda$.

#### The Fundamental Problem

Given:

- The **unperturbed Hamiltonian** $H_0$ with known spectrum: $\{E_n^{(0)}, 
|\psi_n^{(0)}\rangle\}$- A **perturbation** operator $V$- A small parameter $\lambda \ll 1$Find:
- The spectrum of $H(\lambda) = H_0 + \lambda V$ as a power series in $\lambda$:

$$


E_n(\lambda) = E_n^{(0)} + \lambda E_n^{(1)} + \lambda^2 E_n^{(2)} + \ldots

$$ (eq-5a-energy-expansion)

$$ |\psi_n(\lambda)\rangle = 
|\psi_n^{(0)}\rangle + \lambda 
|\psi_n^{(1)}\rangle + \lambda^2 
|\psi_n^{(2)}\rangle + \ldots

$$ (eq-5a-state-expansion)

| Reason 
| Impact 
||--------
|--------
|| **Exact solutions are rare** 
| Most realistic quantum systems (atoms, molecules, condensed matter) have no closed-form solutions. Perturbation theory leverages the few that do. 
|| **Separated energy scales** 
| Physics often occurs at distinct scales (e.g., atomic structure $\approx$ eV, hyperfine splitting $\approx$ mueV). Perturbation theory reveals this hierarchy. 
|| **Physical insight** 
| Systematic expansion reveals *how* states hybridize, *how* levels interact, *why* certain quantities are small. 
|| **Practical calculations** 
| Transition rates, response functions, scattering amplitudes are computed via perturbative series. 
|

### A Pedagogical Toy Model: The Qubit

We consider a **single qubit** (two-level system) with an exact solution. This allows us to compare perturbative results with exact answers and understand where approximations break down.

#### Hamiltonian

$$


H(\lambda) = \hat{\sigma}^z + \lambda \hat{\sigma}^x = \begin{pmatrix} 1 & \lambda \\ \lambda & -1 \end{pmatrix}

$$ (eq-5a-qubit-ham)

where $\hat{\sigma}^x$ and $\hat{\sigma}^z$ are Pauli matrices with $\{
|0\rangle, 
|1\rangle\}$ as the computational basis:

$$


\hat{\sigma}^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}, \quad\hat{\sigma}^x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}

$$ (eq-5a-pauli)

Thus:
- $H_0 = \hat{\sigma}^z$ (unperturbed Hamiltonian)
- $V = \hat{\sigma}^x$ (perturbation)**Physical meaning**: $\hat{\sigma}^z$ favors spin-up (energy $+1$) over spin-down (energy $-1$). The perturbation $\lambda \hat{\sigma}^x$ induces spin flips between the two states.

#### Unperturbed Spectrum

At $\lambda = 0$:

$$


H_0 
|0\rangle = 
|0\rangle, \quad H_0 
|1\rangle = -
|1\rangle

$$ (eq-5a-unperturbed-spectrum)

- State $|0\rangle$ has energy $E_+^{(0)} = +1$ (excited state when labeled by decreasing energy)
- State $|1\rangle$ has energy $E_-^{(0)} = -1$ (ground state)**Notation**: We use subscript $+/-$ to label states in order of decreasing energy, so $E_+^{(0)} > E_-^{(0)}$.

### Exact Solution via Diagonalization

#### Eigenvalue Equation

Solve $\det(H(\lambda) - E\mathbb{I}) = 0$:

$$


\det\begin{pmatrix} 1-E & \lambda \\ \lambda & -1-E \end{pmatrix} = (1-E)(-1-E) - \lambda^2 = 0

$$


$$

-1 - E + E + E^2 - \lambda^2 = 0 \quad \Rightarrow \quad E^2 = 1 + \lambda^2

$$


**Exact eigenvalues**:

$$


E_\pm(\lambda) = \pm\sqrt{1 + \lambda^2}

$$ (eq-5a-exact-energies)

#### Exact Eigenstates

For each eigenvalue, solve $(H - E)\psi = 0$. The normalized eigenstates are:

$$ |\psi_+(\lambda)\rangle = \frac{1}{\sqrt{2(1 + \lambda^2 + \sqrt{1+\lambda^2})}} \begin{pmatrix} 1 + \sqrt{1+\lambda^2} \\ \lambda \end{pmatrix}

$$ (eq-5a-exact-state-plus)

$$ |\psi_-(\lambda)\rangle = \frac{1}{\sqrt{2(1 + \lambda^2 + \sqrt{1+\lambda^2})}} \begin{pmatrix} \lambda \\ -(1 + \sqrt{1+\lambda^2}) \end{pmatrix}

$$ (eq-5a-exact-state-minus)

These expressions are exact for **all** $\lambda \geq 0$.

### Taylor Expansion of Exact Results

Now expand the exact expressions for small $\lambda$. We use:

$$


\sqrt{1 + x} = 1 + \frac{x}{2} - \frac{x^2}{8} + \frac{x^3}{16} - \frac{5x^4}{128} + \ldots, \quad 
|x
| < 1

$$ (eq-5a-sqrt-expansion)

$$


E_\pm(\lambda) = \pm\sqrt{1 + \lambda^2} = \pm\left(1 + \frac{\lambda^2}{2} - \frac{\lambda^4}{8} + \frac{\lambda^6}{16} - \ldots\right)

$$ (eq-5a-energy-series)

**Observation**: The expansion contains only **even powers** of $\lambda$. This is a consequence of symmetry: $H(\lambda)$ is an even function under $\lambda \to -\lambda$ (since the Hamiltonian matrix is real and symmetric), so eigenvalues must also be even functions.Identifying coefficients:

:class: dropdown warning

**Why do first-order corrections vanish?**

You found that $E_+^{(1)} = \langle 0 | \hat{\sigma}^x | 0 \rangle = 0$ and $E_-^{(1)} = \langle 1 | \hat{\sigma}^x | 1 \rangle = 0$. This is *not* a coincidence—it follows from a deeper symmetry.

- The matrix $\hat{\sigma}^x = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$ has zero diagonal elements. What does this tell you about the selection rules for transitions induced by this operator?
- More generally, if the perturbation $V$ commutes with some symmetry of the unperturbed system, which corrections can vanish? 
- Physical consequence: Since $E^{(1)} = 0$, the first physical shift to the energy appears at *second order*. Why is this actually a blessing in practical calculations—what does it mean for experimental accuracy?

:::

$$


E_+^{(0)} = 1, \quad E_+^{(1)} = 0, \quad E_+^{(2)} = \frac{1}{2}, \quad E_+^{(3)} = 0, \quad E_+^{(4)} = -\frac{1}{8}, \ldots

$$ (eq-5a-energy-coeffs)

#### State Series

Expanding the eigenstates requires careful algebra. For the upper state $|\psi_+(\lambda)\rangle$:

$$ |\psi_+(\lambda)\rangle = 
|0\rangle + \frac{\lambda}{2}
|1\rangle - \frac{\lambda^2}{8}
|0\rangle - \frac{3\lambda^3}{16}
|1\rangle + O(\lambda^4)

$$ (eq-5a-state-series-plus)

$$ |\psi_-(\lambda)\rangle = 
|1\rangle - \frac{\lambda}{2}
|0\rangle - \frac{\lambda^2}{8}
|1\rangle + \frac{3\lambda^3}{16}
|0\rangle + O(\lambda^4)

$$ (eq-5a-state-series-minus)

Again, the series contains only even powers in the "unperturbed" part ($|0\rangle$ in $|\psi_+\rangle$) and odd powers in the "admixture" part ($|1\rangle$ in $|\psi_+\rangle$). This reflects the parity of the coupling.

### Power Series and Perturbative Coefficients

Perturbation theory computes the coefficients $E_n^{(k)}$ and $|\psi_n^{(k)}\rangle$ **without solving the system exactly**. The key insight:

$$


E_n^{(k)} = \frac{1}{k!}\left.\frac{\mathrm{d}^k E_n(\lambda)}{d\lambda^k}\right
|_{\lambda=0}

$$ (eq-5a-energy-coeff-deriv)

**Method**: Differentiate the eigenvalue equation $k$ times and evaluate at $\lambda = 0$. The derivatives of $H_0$ and $V$ are:

$$


\frac{\partial H}{\partial \lambda}\bigg
|_{\lambda=0} = V, \quad \frac{\partial^2 H}{\partial \lambda^2}\bigg
|_{\lambda=0} = 0, \quad \ldots

$$ (eq-5a-ham-derivs)

This turns the eigenvalue problem into a **recursive algorithm** that generates all orders of the perturbative expansion.

### Convergence of the Perturbative Series

#### Radius of Convergence

For the qubit model, the exact solution is entire (analytic everywhere), but the perturbative expansion has **finite radius of convergence** $R = 1$. This is determined by the *nearest singularity* in the complex $\lambda$-plane.Setting $E^2 = 1 + \lambda^2 = 0$ gives $\lambda = \pm i$. These are the *branch points* where the two eigenvalue sheets meet. The convergence radius is:

$$


R = 
|\lambda_{\text{singularity}}
| = 1

$$ (eq-5a-convergence-radius)

#### Convergence Criterion



:::{admonition} Discussion
:class: dropdown warning

**When is the perturbation "small"?** 

You have computed that for the qubit, perturbation theory converges for $|\lambda| < 1$. But consider a hydrogen atom in an external magnetic field (Zeeman effect). The unperturbed splitting between $n=1$ and $n=2$ is roughly $13.6\,\mathrm{eV}$, while the magnetic perturbation is $\sim 10^{-4}\,\mathrm{eV}$ for a typical lab field.

- Why does perturbation theory work brilliantly here, even though the gap $\Delta E \gg |V|$ seems huge?
- Now imagine ramping up the magnetic field until $|V| \sim 0.1\,\Delta E$. At what field strength does the perturbation series start to fail?
- How would you know experimentally (e.g., from spectroscopy) that you have entered the regime where perturbation theory breaks down?

The lesson: The *absolute* magnitude of $\lambda$ or $V$ matters far less than the *ratio* $\lambda|V|/\Delta E$.
:::
The perturbative power series converges for $|\lambda
| < 1$. More generally, for any quantum system:

$$


\text{Convergence requires} \quad \lambda \ll \frac{\min(\text{energy gaps})}{\
|V\
|}

$$ (eq-5a-convergence-criterion)

**Physical interpretation**:
- If $\lambda V \sim \Delta E$ (energy gap), the perturbation is no longer "small"; strong mixing occurs.
- For the qubit, the gap is $\Delta E = 2$ and $\
|V\
| = 1$ (largest eigenvalue of $\hat{\sigma}^x$), giving $\lambda \lesssim 2$, consistent with $R = 1$ as a rough bound.

:class: note

Many

 quantum systems have **divergent** or **asymptotic** perturbative series. The series may have zero radius of convergence but still provides excellent approximations at small $\lambda$ if truncated appropriately. The qubit is fortunate: it has a finite, nonzero convergence radius.

:::

### Graphical Insight: Level Crossing and Repulsion

As $\lambda$ increases from 0:

- **At $\lambda = 0$**: Two levels at $E_+ = 1$ and $E_- = -1$ (separated by gap $\Delta E = 2$).
- **For $0 < \lambda < 1$**: Levels curve smoothly; the perturbation couples them, inducing a quadratic splitting (from $E^{(2)}$ terms).
- **At $\lambda = 1$**: Levels reach closest approach, with $\Delta E = 0$ at complex $\lambda = i$.
- **For $\lambda > 1$**: Perturbation series breaks down; exact solution still valid: $E_\pm \approx \pm \lambda$ (grows linearly).The **exact solution** exhibits **level repulsion**: the two branches $E_+(\lambda)$ and $E_-(\lambda)$ never cross for real $\lambda$, though they come close at large $\lambda$. This is a hallmark of avoided crossings.

### Key Takeaways

1. **Perturbation theory** expands eigenvalues/eigenstates as power series in $\lambda$.
2. **The qubit model** is an exactly solvable system where we can compare perturbative approximations with exact results.
3. **Taylor expansion** of the exact solution reveals:   - Energy has only even powers in $\lambda$ (due to symmetry).   - States have even powers in the "unperturbed" component and odd powers in admixtures.   - Coefficients follow a pattern: $E^{(0)}, E^{(2)}, E^{(4)}, \ldots$ (odd orders vanish).
4. **Convergence** is controlled by the proximity to singularities in the complex $\lambda$-plane. For the qubit, $R = 1$.
5. **Physical interpretation**: Perturbation theory is valid when $\lambda V \ll \Delta E$ (coupling strength much smaller than energy gaps).
