## Lecture Notes

### Overview

Sec. 5.1.1 used an exactly solvable toy model as a benchmark. This subsection takes the next step: derive the perturbative coefficients directly from unperturbed data, without exact diagonalization.

We focus on the non-degenerate case and build formulas for first- and second-order energy shifts, first-order state mixing, and the validity condition that tells us when this method breaks down and degenerate perturbation theory is needed.

### Problem Setup

The general perturbative setup $\hat H(\lambda)=\hat H_0+\lambda\hat V$ was introduced in Sec. 5.1.1. Here we focus on the case where $\hat H_0$ has a non-degenerate spectrum,

$$
E_n\neq E_m\text{ for }n\neq m,
$$

so each index $n$ labels a unique unperturbed eigenstate of

$$
\hat H_0|n\rangle=E_n|n\rangle.
$$

Under this non-degeneracy condition, we can use matrix elements

$$
V_{mn}=\langle m|\hat V|n\rangle
$$

to build $E_n(\lambda)$ and $|n(\lambda)\rangle$ order by order.

:::{admonition} Notation Rule
:class: dropdown note

To simplify notation, when $(\lambda)$ is omitted we evaluate at $\lambda=0$:

$$
\hat H\equiv \hat H(0)=\hat H_0,
$$

$$
|n\rangle\equiv |n(0)\rangle,
$$

$$
E_n\equiv E_n(0).
$$

Also, $|\partial_\lambda n\rangle$ means derivative of the **state vector** $|n(\lambda)\rangle$ (not derivative of the integer label $n$):

$$
|\partial_\lambda n\rangle\equiv \left.\frac{\partial}{\partial\lambda}|n(\lambda)\rangle\right|_{\lambda=0}.
$$

If $|n(\lambda)\rangle=\sum_k c_{nk}(\lambda)|k\rangle$, then $|\partial_\lambda n\rangle=\sum_k\left(\partial_\lambda c_{nk}(\lambda)\right)_{\lambda=0}|k\rangle$.
:::

### Hellmann-Feynman Identities (Core Engine)

From the Taylor-expansion viewpoint, perturbation theory is a derivative problem: we need derivatives of energies and states with respect to $\lambda$.

The Hellmann-Feynman identities are exactly the tool we need: they convert those derivatives into matrix elements of $\hat V$, giving a recursive route to higher-order corrections.

:::{admonition} Hellmann-Feynman identities (non-degenerate case)
:class: important

For

$$
\hat H(\lambda)|n(\lambda)\rangle=E_n(\lambda)|n(\lambda)\rangle,
$$

with $\hat H(\lambda)=\hat H_0+\lambda\hat V$, evaluated at $\lambda=0$:

$$
\partial_\lambda E_n = V_{nn},
$$

$$
\langle m|\partial_\lambda n\rangle = \frac{V_{mn}}{E_n-E_m}
\text{ for }m\neq n.
$$

where $V_{mn}=\langle m|\hat V|n\rangle$.
:::

:::{admonition} Derivation
:class: dropdown information

Start from the eigenvalue equation:

$$
\hat H(\lambda)|n(\lambda)\rangle=E_n(\lambda)|n(\lambda)\rangle.
$$

Differentiate with respect to $\lambda$:

$$
\partial_\lambda\hat H|n\rangle + \hat H|\partial_\lambda n\rangle
= \partial_\lambda E_n|n\rangle + E_n|\partial_\lambda n\rangle.
$$

Left-multiply by $\langle m|$:

$$
\langle m|\partial_\lambda\hat H|n\rangle + \langle m|\hat H|\partial_\lambda n\rangle
= \partial_\lambda E_n\langle m|n\rangle + E_n\langle m|\partial_\lambda n\rangle.
$$

Use $\langle m|\hat H=E_m\langle m|$, orthonormality $\langle m|n\rangle=\delta_{mn}$, and $\partial_\lambda\hat H=\hat V$:

$$
V_{mn}=\partial_\lambda E_n\,\delta_{mn}+(E_n-E_m)\langle m|\partial_\lambda n\rangle.
$$

Now separate two cases:

- If $m=n$:

$$
V_{nn}=\partial_\lambda E_n.
$$

- If $m\neq n$:

$$
V_{mn}=(E_n-E_m)\langle m|\partial_\lambda n\rangle,
$$

so

$$
\langle m|\partial_\lambda n\rangle=\frac{V_{mn}}{E_n-E_m}.
$$

This is the coupling-over-gap structure that controls state mixing.
:::

With these identities in hand, we can now compute energy and state corrections order by order.

### Energy Corrections

Now use the Taylor expansion of the eigenenergy around $\lambda=0$. Once $\partial_\lambda E_n$ and $\partial_\lambda^2E_n$ are known, the perturbative coefficients follow immediately.

:::{admonition} Energy expansion up to second order
:class: important

$$
E_n(\lambda)=E_n+(\partial_\lambda E_n)\lambda+\frac{1}{2}(\partial_\lambda^2E_n)\lambda^2+O(\lambda^3).
$$

Using Hellmann-Feynman identities:

$$
\partial_\lambda E_n=V_{nn},
$$

$$
\partial_\lambda^2E_n=2\sum_{m\neq n}\frac{V_{nm}V_{mn}}{E_n-E_m}.
$$

So

$$
E_n(\lambda)=E_n+V_{nn}\lambda+\sum_{m\neq n}\frac{|V_{mn}|^2}{E_n-E_m}\lambda^2+O(\lambda^3).
$$
:::

:::{admonition} Derivation: first-order energy derivative
:class: dropdown information

From the first Hellmann-Feynman identity (the $m=n$ case):

$$
\partial_\lambda E_n=V_{nn}=\langle n|\hat V|n\rangle.
$$

This is exactly the first derivative needed in the Taylor series.
:::

:::{admonition} Derivation: second-order energy derivative
:class: dropdown information

Start from

$$
\partial_\lambda E_n=\langle n|\hat V|n\rangle.
$$

Differentiate once more:

$$
\partial_\lambda^2E_n
=\langle\partial_\lambda n|\hat V|n\rangle+\langle n|\hat V|\partial_\lambda n\rangle.
$$

Insert resolution of identity and keep $m\neq n$ terms:

$$
\partial_\lambda^2E_n
=\sum_{m\neq n}\left(\langle\partial_\lambda n|m\rangle V_{mn}+V_{nm}\langle m|\partial_\lambda n\rangle\right).
$$

Use

$$
\langle m|\partial_\lambda n\rangle=\frac{V_{mn}}{E_n-E_m},
$$

$$
\langle\partial_\lambda n|m\rangle=\frac{V_{nm}}{E_n-E_m}.
$$

then

$$
\partial_\lambda^2E_n=2\sum_{m\neq n}\frac{V_{nm}V_{mn}}{E_n-E_m}.
$$
:::

### State Corrections

Apply the same Taylor logic to states:

:::{admonition} State expansion and first-order correction
:class: important

$$
|n(\lambda)\rangle=|n\rangle+|\partial_\lambda n\rangle\lambda+O(\lambda^2).
$$

Using the second Hellmann-Feynman identity:

$$
|\partial_\lambda n\rangle=\sum_{m\neq n}|m\rangle\frac{V_{mn}}{E_n-E_m},
$$

so

$$
|n(\lambda)\rangle=|n\rangle+\sum_{m\neq n}|m\rangle\frac{V_{mn}}{E_n-E_m}\lambda+O(\lambda^2).
$$
:::

:::{admonition} Derivation: first-order state derivative
:class: dropdown information

Expand the derivative vector in the unperturbed basis:

$$
|\partial_\lambda n\rangle=\sum_m|m\rangle\langle m|\partial_\lambda n\rangle.
$$

For $m=n$, this term can be set to zero by phase convention. For $m\neq n$, use

$$
\langle m|\partial_\lambda n\rangle=\frac{V_{mn}}{E_n-E_m}.
$$

Therefore

$$
|\partial_\lambda n\rangle=\sum_{m\neq n}|m\rangle\frac{V_{mn}}{E_n-E_m}.
$$
:::

This makes the physical meaning transparent: mixing is stronger for larger coupling and smaller energy gap.

### Physical Intuition and Validity

- **Diagonal matrix elements** $V_{nn}$ shift energies at first order.
- **Off-diagonal matrix elements** $V_{mn}$ mix states and generate second-order shifts.
- **Level repulsion**: virtual transitions push levels apart.
- **Breakdown criterion**: if $|E_n-E_m|$ becomes comparable to $|V_{mn}|$, denominators become large and non-degenerate perturbation theory loses validity.

Then we must switch to **degenerate perturbation theory** (Sec. 5.1.3).

:::{admonition} Example: check against the qubit toy model
:class: dropdown example

For $\hat H=\hat Z+\lambda \hat X$ with $E_+=+1$, $E_-=-1$ and $V_{++}=V_{--}=0$:

$$
E_+^{(1)}=0,
$$

$$
E_+^{(2)}=\frac{|\langle -|\hat X|+\rangle|^2}{E_+-E_-}=\frac{1}{2},
$$

so

$$
E_+(\lambda)=1+\frac{\lambda^2}{2}+\cdots,
$$

consistent with

$$
\sqrt{1+\lambda^2}=1+\frac{\lambda^2}{2}-\frac{\lambda^4}{8}+\cdots.
$$

Also,

$$
|+^{(1)}\rangle = \frac{1}{2}|-\rangle,
$$

which matches the exact eigenvector expansion at small $\lambda$.
:::

:::{admonition} Discussion: what does divergence mean?
:class: dropdown tip

When $E_n-E_m\to 0$, formulas like $V_{mn}/(E_n-E_m)$ diverge. Is this a physical singularity, or a sign that our basis choice is no longer appropriate? Explain how block-diagonalization inside the nearly degenerate subspace resolves this issue.
:::

### Summary

- Non-degenerate perturbation theory is an **iterative derivative algorithm** built from Hellmann-Feynman identities.
- First-order energy comes from diagonal matrix elements: $E_n^{(1)}=V_{nn}$.
- First-order state mixing is coupling over gap: $|n^{(1)}\rangle=\sum_{m\neq n}|m\rangle V_{mn}/(E_n-E_m)$.
- Second-order energy is a sum over virtual processes: $E_n^{(2)}=\sum_{m\neq n}|V_{mn}|^2/(E_n-E_m)$.
- Near degeneracy is not a failure of QM; it signals a change of method (degenerate perturbation theory).
