## Lecture Notes

### Overview

Non-degenerate perturbation theory provides a systematic expansion of eigenvalues and eigenstates in powers of a small parameter $\lambda$, for the Hamiltonian $H(\lambda) = H_0 + \lambda V$. The **Hellmann-Feynman theorem** is the organizing principle: the exact rate at which an eigenvalue changes with $\lambda$ equals the expectation value of $V$ in the exact eigenstate.

### Hellmann-Feynman Theorem

:::{admonition} Hellmann-Feynman Theorem
:class: important

For a parameter-dependent Hamiltonian $H(\lambda)$ with eigenvalue $E_n(\lambda)$ and normalized eigenstate $\vert\psi_n(\lambda)\rangle$:

$$
\frac{\mathrm{d}E_n}{\mathrm{d}\lambda} = \langle \psi_n(\lambda) \vert \frac{\partial H}{\partial \lambda} \vert \psi_n(\lambda) \rangle
$$ (eq-5b-hellmann-feynman)

At $\lambda = 0$ with $\partial H/\partial\lambda = V$, this immediately gives the first-order energy:

$$
E_n^{(1)} = \frac{\mathrm{d}E_n}{\mathrm{d}\lambda}\bigg\vert_{\lambda=0} = \langle \psi_n^{(0)} \vert V \vert \psi_n^{(0)} \rangle
$$

**Requires**: adiabatic continuity — eigenstates can be tracked continuously as $\lambda$ varies (no level crossings).
:::

### First-Order Energy Correction

:::{admonition} First-Order Energy
:class: important

$$
\boxed{E_n^{(1)} = \langle \psi_n^{(0)} \vert V \vert \psi_n^{(0)} \rangle}
$$ (eq-5b-E1-result)

The first-order shift is the **diagonal matrix element** of $V$ in the unperturbed basis — the expectation value of the perturbation in the unperturbed state.
:::

If $V$ has no diagonal elements (e.g., $\hat{\sigma}^x$ in the computational basis), the first-order correction vanishes and the leading effect is second-order.

:::{admonition} Derivation
:class: dropdown information

Expand $H\vert\psi_n\rangle = E_n\vert\psi_n\rangle$ to order $\lambda^1$:

$$
V\vert\psi_n^{(0)}\rangle + H_0\vert\psi_n^{(1)}\rangle = E_n^{(0)} \vert\psi_n^{(1)}\rangle + E_n^{(1)} \vert\psi_n^{(0)}\rangle
$$

Project onto $\langle \psi_n^{(0)}\vert$ and use $\langle \psi_n^{(0)}\vert H_0 = E_n^{(0)}\langle \psi_n^{(0)}\vert$ plus the normalization choice $\langle \psi_n^{(0)}\vert\psi_n^{(1)}\rangle = 0$:

$$
\langle \psi_n^{(0)} \vert V \vert \psi_n^{(0)} \rangle = E_n^{(1)}
$$

:::

### First-Order State Correction

:::{admonition} First-Order State
:class: important

$$
\boxed{\vert\psi_n^{(1)}\rangle = \sum_{m \neq n} \frac{\langle \psi_m^{(0)} \vert V \vert \psi_n^{(0)} \rangle}{E_n^{(0)} - E_m^{(0)}} \vert\psi_m^{(0)}\rangle}
$$ (eq-5b-psi1-result)

The admixture of state $m$ is proportional to the **coupling** $\langle m\vert V\vert n\rangle$ and inversely proportional to the **energy gap** $E_n^{(0)} - E_m^{(0)}$. States mix strongest when they are close in energy and strongly coupled.
:::

:::{admonition} Derivation
:class: dropdown information

Expand $\vert\psi_n^{(1)}\rangle = \sum_{m\neq n} c_m^{(1)}\vert\psi_m^{(0)}\rangle$ (excluding $m=n$ by orthogonality). Insert into the order-$\lambda^1$ equation and project onto $\langle\psi_m^{(0)}\vert$:

$$
\langle \psi_m^{(0)} \vert V \vert \psi_n^{(0)} \rangle = c_m^{(1)}(E_m^{(0)} - E_n^{(0)})
$$

Solving: $c_m^{(1)} = \langle \psi_m^{(0)} \vert V \vert \psi_n^{(0)} \rangle / (E_n^{(0)} - E_m^{(0)})$.
:::

### Second-Order Energy Correction

:::{admonition} Second-Order Energy
:class: important

$$
\boxed{E_n^{(2)} = \sum_{m \neq n} \frac{\vert\langle \psi_m^{(0)} \vert V \vert \psi_n^{(0)} \rangle\vert^2}{E_n^{(0)} - E_m^{(0)}}}
$$ (eq-5b-E2-result)

- **Ground state**: all denominators are negative, so $E_0^{(2)} \leq 0$ — perturbations always lower the ground state energy (consistent with the variational principle).
- **Level repulsion**: nearby levels push each other apart through virtual transitions.
- **Convergence**: requires $\lambda \vert\langle m\vert V \vert n\rangle\vert \ll \vert E_n^{(0)} - E_m^{(0)} \vert$ for all $m$.
:::

:::{admonition} Derivation
:class: dropdown information

From the order-$\lambda^2$ equation, $E_n^{(2)} = \langle\psi_n^{(0)}\vert V\vert\psi_n^{(1)}\rangle$. Substituting the first-order state and using Hermiticity of $V$:

$$
E_n^{(2)} = \sum_{m\neq n} \frac{\langle\psi_n^{(0)}\vert V\vert\psi_m^{(0)}\rangle\langle\psi_m^{(0)}\vert V\vert\psi_n^{(0)}\rangle}{E_n^{(0)} - E_m^{(0)}} = \sum_{m\neq n}\frac{\vert\langle m^{(0)}\vert V\vert n^{(0)}\rangle\vert^2}{E_n^{(0)} - E_m^{(0)}}
$$

:::

### Physical Intuitions

The perturbation series has a clean physical interpretation at each order:

- **First-order energy** = expectation value of the perturbation (average energy shift in the unperturbed state)
- **First-order state** = hybridization with other levels, weighted by coupling/gap
- **Second-order energy** = virtual transitions to intermediate states and back; always lowers the ground state
- **Breakdown**: when $E_n^{(0)} - E_m^{(0)} \to 0$ (degeneracy), denominators diverge → signal to switch to degenerate perturbation theory (\S5.1.3)

:::{admonition} Example: Verification Against the Qubit Model
:class: dropdown example

**Problem.** Verify perturbation theory on the qubit $H(\lambda) = \hat{\sigma}^z + \lambda\hat{\sigma}^x$ from \S5.1.1.

**Solution.** Unperturbed: $E_+^{(0)} = +1$ ($\vert 0\rangle$), $E_-^{(0)} = -1$ ($\vert 1\rangle$). Since $\langle 0\vert\hat{\sigma}^x\vert 0\rangle = 0$, the first-order energy vanishes: $E_+^{(1)} = 0$. The second-order correction:

$$
E_+^{(2)} = \frac{\vert\langle 1\vert\hat{\sigma}^x\vert 0\rangle\vert^2}{E_+^{(0)} - E_-^{(0)}} = \frac{1}{2}
$$

So $E_+(\lambda) = 1 + \tfrac{1}{2}\lambda^2 + O(\lambda^3)$, matching the exact result $\sqrt{1+\lambda^2} = 1 + \tfrac{1}{2}\lambda^2 - \tfrac{1}{8}\lambda^4 + \cdots$. $\checkmark$

The first-order state: $\vert\psi_+^{(1)}\rangle = \tfrac{1}{2}\vert 1\rangle$, so $\vert\psi_+(\lambda)\rangle \approx \vert 0\rangle + \tfrac{\lambda}{2}\vert 1\rangle$, again matching the exact eigenvector expansion. $\checkmark$
:::

### Summary

- The **Hellmann-Feynman theorem** $\mathrm{d}E_n/\mathrm{d}\lambda = \langle\psi_n\vert V\vert\psi_n\rangle$ is the organizing principle: energy derivatives are expectation values.
- **First-order energy**: $E_n^{(1)} = \langle n^{(0)}\vert V\vert n^{(0)}\rangle$ — expectation value of perturbation.
- **First-order state**: admixture of other levels, weighted by coupling over energy gap.
- **Second-order energy**: virtual transitions; always lowers the ground state; series diverges at degeneracy.
