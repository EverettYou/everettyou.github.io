## Lecture Notes

### Overview

This lesson applies first-order time-dependent perturbation theory to four key physical systems: harmonic driving that reveals resonance and Rabi oscillations, electric dipole transitions with selection rules, decay and linewidth from finite lifetime, and linear response theory leading to the quantum Hall conductance—a remarkable exact result from perturbation theory.

### Harmonic Perturbation

The simplest time-dependent perturbation is a harmonic (oscillating) drive:

$$
V(t) = V_0 \cos(\omega t)
$$

We can rewrite this as a sum of positive and negative frequency components:

$$
V(t) = \frac{V_0}{2} \left( \mathrm{e}^{\mathrm{i}\omega t} + \mathrm{e}^{-\mathrm{i}\omega t} \right)
$$

The transition amplitude from state $\vert i \rangle$ to $\vert f \rangle$ picks up resonance when the driving frequency matches the energy gap:

$$
c_f^{(1)}(t) \propto \langle f \vert V_0 \vert i \rangle \left[ \frac{\mathrm{e}^{\mathrm{i}(\omega - \omega_{fi})t} - 1}{\omega - \omega_{fi}} + \frac{\mathrm{e}^{-\mathrm{i}(\omega + \omega_{fi})t} - 1}{\omega + \omega_{fi}} \right]
$$

where $\omega_{fi} = (E_f - E_i)/\hbar$. The first term dominates when $\omega \approx \omega_{fi}$ (absorption), and the second when $\omega \approx -\omega_{fi}$ (stimulated emission). This is the rotating wave approximation: we keep only the resonant term and drop the rapidly oscillating counter-rotating term.

Near resonance, the transition probability oscillates at the **Rabi frequency**:

:::{admonition} Rabi Oscillations
:class: important

In the rotating frame near resonance, the transition probability is

$$
P(t) = \sin^2\left(\frac{\Omega_R t}{2}\right) \quad \text{where} \quad \Omega_R = \frac{\vert V_{fi}\vert}{\hbar}
$$

is the Rabi frequency. Population coherently oscillates between the two states at rate $\Omega_R$. A $\pi$-pulse ($\Omega_R t = \pi$) completely flips the state.

:::

A key application: **scanning the driving frequency $\omega$ probes the energy level spacing**. Each resonance reveals $E_f - E_i$, forming the basis of spectroscopy.

This resonant behavior provides a framework for understanding which transitions are actually possible in real systems. Specifically, not all transitions allowed by energy conservation are allowed by the quantum mechanical structure of the wavefunctions themselves—selection rules determine this.

### Electric Dipole Transitions

When light couples to matter, the interaction is dominated by the dipole term:

$$
V(t) = -\boldsymbol{d} \cdot \boldsymbol{E}(t) = -(-e\boldsymbol{r}) \cdot \boldsymbol{E}(t) = e \boldsymbol{r} \cdot \boldsymbol{E}(t)
$$

where $\boldsymbol{d} = -e\boldsymbol{r}$ is the electric dipole moment. The transition amplitude depends on the dipole matrix element $\langle f \vert \boldsymbol{r} \vert i \rangle$.

Not all transitions are allowed. The parity and angular momentum structure of wavefunctions impose **selection rules**. For hydrogen (or any single-electron atom):

:::{admonition} Selection Rules for Electric Dipole Transitions
:class: note

For a transition $n\ell m_\ell \to n'\ell' m_{\ell}'$:
- $\Delta \ell = \pm 1$ (change parity)
- $\Delta m_\ell = 0, \pm 1$ (depends on light polarization)
- $\Delta S = 0$ (spin unchanged in non-relativistic theory)

Transitions violating these are forbidden and have suppressed amplitude. States that can only decay via forbidden transitions become **metastable** (long-lived).

:::

For example, the 2s state of hydrogen cannot decay to 1s by electric dipole radiation (both $\ell = 0$, so $\Delta \ell = 0$), making 2s metastable. The 2p state, however, can decay resonantly.

### Lifetime and Linewidth

An excited state coupled to a continuum of final states (e.g., radiating photons) decays exponentially. From Fermi's golden rule applied to the emission process, the decay rate is

$$
\Gamma = \frac{2\pi}{\hbar} \sum_f \vert\langle f \vert V \vert i \rangle\vert^2 \delta(E_i - E_f - \hbar\omega_{\text{photon}})
$$

The inverse is the **lifetime**:

:::{admonition} Lifetime and Natural Linewidth
:class: important

$$
\tau = \frac{1}{\Gamma}
$$

The energy-time uncertainty principle implies that a state with finite lifetime $\tau$ has an intrinsic energy uncertainty:

$$
\Delta E \sim \hbar \Gamma = \frac{\hbar}{\tau}
$$

This gives a **natural linewidth** in the emission spectrum. The lineshape is Lorentzian, centered at the transition energy $E_i - E_f$ with width $\Gamma$.

:::

For instance, the 2p state of hydrogen has $\tau \sim 1.6 \times 10^{-9}$ s (lifetime), corresponding to $\Gamma \sim 0.1$ eV (linewidth).

So far we have focused on single isolated atoms or simple two-level systems. The same first-order machinery extends to many-body response functions: how a macroscopic observable (e.g. electric current) responds to a weak external field. The result is **linear-response theory**, whose central object is the Kubo formula.

### Linear Response and the Kubo Formula

Consider non-interacting electrons in a single-particle Hamiltonian $\hat{H}_0$ with eigenstates $\{|\alpha\rangle\}$ and energies $\{E_\alpha\}$. At zero temperature every state below the Fermi level is occupied, every state above is empty. The single-particle current operator is $\hat{\boldsymbol{j}} = q\hat{\boldsymbol{v}} = q\hat{\boldsymbol{\pi}}/m$ (charge times velocity), and the **macroscopic current density** is its expectation value summed over occupied single-particle states, per unit area:

$$
\boldsymbol{j} \;\equiv\; \frac{1}{A}\sum_{\alpha\in\mathrm{occ}}\langle\alpha|\hat{\boldsymbol{j}}|\alpha\rangle.
$$ (eq-current-density-def)

A weak electric field $\boldsymbol{E}(t)$ induces a current linear in the field,

$$
j_a(t) = \sigma_{ab}(\omega)\,E_b(t),
$$

defining the conductivity tensor $\sigma_{ab}(\omega)$. We now compute it from first-order time-dependent perturbation theory (§5.2.1).

**Coupling to the field via the vector potential.** A spatially uniform, oscillatory $\boldsymbol{E}(t) = \boldsymbol{E}\,\mathrm{e}^{-\mathrm{i}(\omega+\mathrm{i}0^+)t}$ (adiabatically turned on from $t = -\infty$, hence the $0^+$ regulator) is implemented by a vector-potential perturbation

$$
\delta\boldsymbol{A}(t) = \frac{\mathrm{i}\boldsymbol{E}}{\omega+\mathrm{i}0^+}\,\mathrm{e}^{-\mathrm{i}(\omega+\mathrm{i}0^+)t}, \qquad \boldsymbol{E}(t) = -\partial_t\,\delta\boldsymbol{A}(t).
$$

Expanding the minimal-coupling Hamiltonian $\hat{H} = (\hat{\boldsymbol{p}} - q\boldsymbol{A})^2/(2m) + V$ to first order in $\delta\boldsymbol{A}$ gives the perturbation

$$
\delta\hat{H}(t) = \frac{\partial\hat{H}}{\partial\boldsymbol{A}}\cdot\delta\boldsymbol{A}(t) = -\hat{\boldsymbol{j}}\cdot\delta\boldsymbol{A}(t),
$$ (eq-kubo-perturbation)

(the partial derivative is taken at fixed canonical $\hat{\boldsymbol{p}}$).

**Result.** Applying first-order perturbation theory state-by-state and summing over occupied states yields the **Kubo formula** for the Hall (transverse) conductivity in the DC limit $\omega\to 0$:

:::{admonition} Kubo formula
:class: important

$$
\sigma_{xy} = \frac{\mathrm{i}\hbar}{A}\sum_{\alpha\in\mathrm{occ}}\sum_{\beta\notin\mathrm{occ}}\frac{\langle\alpha|\hat{j}_x|\beta\rangle\langle\beta|\hat{j}_y|\alpha\rangle - \langle\alpha|\hat{j}_y|\beta\rangle\langle\beta|\hat{j}_x|\alpha\rangle}{(E_\beta - E_\alpha)^2}.
$$ (eq-kubo-hall-523)

The sum runs over occupied single-particle states $|\alpha\rangle$ and empty states $|\beta\rangle$ (no Fermi-Dirac function needed at $T=0$). The numerator is the antisymmetric current-matrix-element combination; the denominator is the squared energy gap $E_\beta - E_\alpha > 0$ across the Fermi level. The same formula appears in §4.3.3 ({eq}`eq-kubo-hall`), where it is evaluated explicitly on Landau-level eigenstates.
:::

:::{admonition} Derivation: Kubo formula
:class: dropdown information

Each electron occupies a state $|\alpha\rangle$ that evolves independently under $\hat{H}_0 + \delta\hat{H}(t)$. We compute the response state-by-state, then sum over occupied states and divide by $A$ to assemble the macroscopic current density $\boldsymbol{j}$ via {eq}`eq-current-density-def`.

**Step 1 — First-order response of one occupied state.**

Each occupied single-particle state $|\alpha\rangle$ evolves under the interaction-picture operator $\hat{U}_I(t)$ from §5.2.1. The expectation value of the single-particle current in that state is

$$
\langle\hat{j}_a(t)\rangle_\alpha \equiv \langle\alpha|\hat{U}_I^\dagger(t)\,\hat{j}_a^I(t)\,\hat{U}_I(t)|\alpha\rangle.
$$

Expanding to first order in $\delta\hat{H}$ gives the standard first-order Kubo identity,

$$
\langle\hat{j}_a(t)\rangle_\alpha = \langle\alpha|\hat{j}_a^I(t)|\alpha\rangle - \frac{\mathrm{i}}{\hbar}\int_{-\infty}^{t}\!\mathrm{d}t'\,\langle\alpha|[\hat{j}_a^I(t),\delta\hat{H}_I(t')]|\alpha\rangle,
$$

with $\hat{j}_a^I(t) = \mathrm{e}^{\mathrm{i}\hat{H}_0 t/\hbar}\hat{j}_a\,\mathrm{e}^{-\mathrm{i}\hat{H}_0 t/\hbar}$.

**Step 2 — Sum over occupied states; assemble the current density.**

Substitute $\delta\hat{H}_I(t') = -\hat{j}_b^I(t')\,\delta A_b(t')$, sum over occupied states, and divide by $A$. The zeroth-order term $\sum_{\alpha\in\mathrm{occ}}\langle\alpha|\hat{j}_a|\alpha\rangle = 0$ vanishes (no current without a driving field), leaving

$$
j_a(t) = \frac{\mathrm{i}}{\hbar A}\!\int_{-\infty}^{t}\!\mathrm{d}t'\,\delta A_b(t')\sum_{\alpha\in\mathrm{occ}}\langle\alpha|[\hat{j}_a^I(t),\hat{j}_b^I(t')]|\alpha\rangle.
$$

**Step 3 — Substitute the harmonic field.**

Using $\delta A_b(t') = (\mathrm{i}E_b/(\omega+\mathrm{i}0^+))\,\mathrm{e}^{-\mathrm{i}(\omega+\mathrm{i}0^+)t'}$ and the change of variable $\tau = t-t'$ (with time-translation invariance of the unperturbed system), the result $j_a(t) = \sigma_{ab}(\omega)\,E_b\,\mathrm{e}^{-\mathrm{i}(\omega+\mathrm{i}0^+)t}$ defines

$$
\sigma_{ab}(\omega) = -\frac{1}{\hbar(\omega+\mathrm{i}0^+) A}\int_0^{\infty}\!\mathrm{d}\tau\,\mathrm{e}^{\mathrm{i}(\omega+\mathrm{i}0^+)\tau}\sum_{\alpha\in\mathrm{occ}}\langle\alpha|[\hat{j}_a^I(\tau),\hat{j}_b^I(0)]|\alpha\rangle.
$$

**Step 4 — Insert eigenbasis and evaluate the time integral.**

Insert $\sum_\gamma|\gamma\rangle\langle\gamma|$ between the two single-particle current operators and use $\hat{j}_a^I(\tau)|\gamma\rangle = \mathrm{e}^{\mathrm{i}\omega_{\gamma\alpha}\tau}\hat{j}_a|\gamma\rangle$ on each $\langle\alpha|\cdots|\alpha\rangle$ matrix element (with $\omega_{\gamma\alpha} = (E_\gamma-E_\alpha)/\hbar$),

$$
\sum_{\alpha\in\mathrm{occ}}\langle\alpha|[\hat{j}_a^I(\tau),\hat{j}_b^I(0)]|\alpha\rangle = \sum_{\alpha\in\mathrm{occ}}\sum_\gamma\Big(j^a_{\alpha\gamma}j^b_{\gamma\alpha}\,\mathrm{e}^{\mathrm{i}\omega_{\gamma\alpha}\tau} - j^b_{\alpha\gamma}j^a_{\gamma\alpha}\,\mathrm{e}^{-\mathrm{i}\omega_{\gamma\alpha}\tau}\Big),
$$

with $j^a_{\alpha\gamma} \equiv \langle\alpha|\hat{j}_a|\gamma\rangle$ a single-particle matrix element. The $\tau$ integral gives $\int_0^\infty\!\mathrm{d}\tau\,\mathrm{e}^{\mathrm{i}(\omega+\mathrm{i}0^+ \pm\omega_{\gamma\alpha})\tau} = \mathrm{i}/(\omega+\mathrm{i}0^+\pm\omega_{\gamma\alpha})$.

**Step 5 — DC limit.**

For $\gamma = \alpha$ (intra-band), the integrand is finite but the prefactor $1/\omega$ produces a pole — the Drude weight, which is absent for a gapped insulator and irrelevant to the antisymmetric Hall part. For $\gamma\neq\alpha$ (inter-band), expand $1/(\omega\pm\omega_{\gamma\alpha}) = \pm 1/\omega_{\gamma\alpha} - \omega/\omega_{\gamma\alpha}^2 + O(\omega^2)$. The $\pm 1/\omega_{\gamma\alpha}$ piece, divided by $\omega$, gives the symmetric (longitudinal) divergence that cancels in clean systems with current conservation. The $-\omega/\omega_{\gamma\alpha}^2$ piece gives a finite **antisymmetric** contribution. Restricting $\gamma\notin\mathrm{occ}$ — terms with $\gamma\in\mathrm{occ}$ pair-cancel because both $\alpha$ and $\gamma$ run over the occupied set — yields

$$
\sigma_{xy} = \frac{\mathrm{i}\hbar}{A}\sum_{\alpha\in\mathrm{occ}}\sum_{\beta\notin\mathrm{occ}}\frac{\langle\alpha|\hat{j}_x|\beta\rangle\langle\beta|\hat{j}_y|\alpha\rangle - \langle\alpha|\hat{j}_y|\beta\rangle\langle\beta|\hat{j}_x|\alpha\rangle}{(E_\beta-E_\alpha)^2},
$$

the Kubo formula above. All matrix elements are single-particle ($\hat{j}$); the $1/A$ traces back to the current-density definition {eq}`eq-current-density-def`.
:::

**Physical content.** The Kubo formula expresses transport coefficients in terms of current-current matrix elements between filled and empty single-particle states. It is the perturbative expression of the *fluctuation-dissipation* relation: response coefficients are determined by ground-state correlation functions. The denominator $(E_\beta - E_\alpha)^2$ shows that low-energy interband excitations dominate.

### Quantum Hall Conductance

For 2D electrons in a perpendicular magnetic field with $\nu$ completely filled Landau levels, the Kubo formula evaluates to the exact integer-quantized result $\sigma_{xy} = \nu\,e^2/h$, independent of disorder, geometry, or microscopic details (provided the bulk gap stays open). The explicit Landau-level evaluation, and the complementary topological (charge-pumping) interpretation of the same quantization, are in **§4.3.3**.

:::{admonition} Discussion: Why is the Hall conductance exactly quantized?
:class: dropdown tip

The Kubo formula is a first-order perturbative result, yet it gives an *exact* integer for filled Landau levels. Some questions to chew on:

- The Kubo expression sums over all occupied/empty pairs and weights each by $1/(E_\beta-E_\alpha)^2$. Why does the answer become a pure integer when the occupied set is a *full* Landau level?
- Disorder localizes most states inside each broadened Landau level. Why does the Kubo sum, which references *all* states, not move off the integer plateau?
- Compare the Kubo derivation (a perturbation-theory calculation) with the Laughlin charge-pumping argument in §4.3.3 (a topological argument). What does it mean that two completely different methods give the same exact answer?
- If you slowly deform the Hamiltonian without closing the bulk gap, can $\sigma_{xy}$ change continuously? What if you close the gap?
:::

### Summary

- **Harmonic driving** at frequency $\omega$ induces resonant transitions when $\omega = \omega_{fi}$; Rabi oscillations probe the coupling strength.

- **Dipole selection rules** ($\Delta \ell = \pm 1$, etc.) determine which transitions are allowed; metastable states bypass these rules.

- **Finite lifetime** $\tau = 1/\Gamma$ produces natural linewidth $\Delta E \sim \hbar/\tau$; spectral lines have intrinsic width.

- **Linear response and the Kubo formula:** first-order time-dependent perturbation theory in the vector-potential perturbation $\delta\hat{H}=-\hat{\boldsymbol{j}}\cdot\delta\boldsymbol{A}$ yields a zero-temperature, occupied-vs-empty form for $\sigma_{xy}$. Evaluated on $\nu$ filled Landau levels (§4.3.3), it gives the exact quantized $\sigma_{xy}=\nu e^2/h$.
