## Lecture Notes

### Overview

The core method is unchanged: use first-order time-dependent perturbation theory to compute transition amplitudes, then interpret the result in different physical settings.

This subsection follows one clean thread:
1. **Driven two-level dynamics** (resonance and Rabi oscillations),
2. **Atomic transitions** (dipole matrix elements and selection rules),
3. **Continuum final states** (golden-rule rate, lifetime, linewidth),
4. **Many-body response** (Kubo formula and quantized Hall conductance).

The key lecture idea is that one perturbative formalism supports both spectroscopy and transport.

### Application 1: Harmonic Driving and Resonance

Consider a monochromatic perturbation

$$
V(t)=V_0\cos(\omega t)=\frac{V_0}{2}\left(\mathrm{e}^{\mathrm{i}\omega t}+\mathrm{e}^{-\mathrm{i}\omega t}\right).
$$

For a transition $|i\rangle\to|f\rangle$ with $\omega_{fi}=(E_f-E_i)/\hbar$, first-order amplitudes contain denominators $(\omega-\omega_{fi})$ and $(\omega+\omega_{fi})$. Near resonance $\omega\approx\omega_{fi}$, the resonant term dominates (rotating-wave approximation).

:::{admonition} Resonance and Rabi scale
:class: important

Near resonance in an effective two-level description,

$$
P_{i\to f}(t)=\sin^2\!\left(\frac{\Omega_R t}{2}\right),\qquad
\Omega_R=\frac{|V_{fi}|}{\hbar}.
$$

So spectroscopy reads off level spacings from resonance positions, while coherent control uses pulse area ($\pi$ pulse, $\pi/2$ pulse, etc.).
:::

### Application 2: Electric Dipole Transitions and Selection Rules

For light-matter coupling in the dipole approximation,

$$
V(t)=-\boldsymbol{d}\cdot\boldsymbol{E}(t)=e\,\boldsymbol{r}\cdot\boldsymbol{E}(t),
$$

so transition strength is controlled by $\langle f|\boldsymbol{r}|i\rangle$.

:::{admonition} Dipole selection rules (single-electron atom)
:class: note

For $n\ell m_\ell\to n'\ell' m_\ell'$:
- $\Delta \ell=\pm1$,
- $\Delta m_\ell=0,\pm1$ (polarization dependent),
- $\Delta S=0$ in non-relativistic electric-dipole transitions.

Hence some channels are strongly allowed while others are suppressed (metastable states).
:::

### Application 3: Golden Rule, Lifetime, and Linewidth

When a discrete initial state couples to many final states, first-order perturbation theory gives a transition **rate** rather than coherent oscillation:

$$
\Gamma=\frac{2\pi}{\hbar}\sum_f|\langle f|V|i\rangle|^2\delta(E_f-E_i).
$$

This is Fermi's golden rule. The same quantity sets spectroscopic broadening.

:::{admonition} Lifetime-linewidth link
:class: important

$$
\tau=\frac{1}{\Gamma},\qquad \Delta E\sim\hbar\Gamma=\frac{\hbar}{\tau}.
$$

Finite lifetime implies a Lorentzian line with intrinsic (natural) width.
:::

:::

### Application 4: Linear Response and Kubo Formula

Now apply the same perturbative logic to transport. For a weak uniform electric field,

$$
\boldsymbol{E}(t)=\boldsymbol{E}\,\mathrm{e}^{-\mathrm{i}(\omega+\mathrm{i}0^+)t},
$$

introduce

$$
\delta\boldsymbol{A}(t)=\frac{\mathrm{i}\boldsymbol{E}}{\omega+\mathrm{i}0^+}\,\mathrm{e}^{-\mathrm{i}(\omega+\mathrm{i}0^+)t},
\qquad \boldsymbol{E}(t)=-\partial_t\delta\boldsymbol{A}(t),
$$

and from minimal coupling obtain

$$
\delta\hat{H}(t)=-\hat{\boldsymbol{j}}\cdot\delta\boldsymbol{A}(t).
$$ (eq-kubo-perturbation)

First-order response of current to this perturbation gives the Hall conductivity.

:::{admonition} Kubo formula (zero temperature)
:class: important

$$
\sigma_{xy}=\frac{\mathrm{i}\hbar}{A}
\sum_{\alpha\in\mathrm{occ}}\sum_{\beta\notin\mathrm{occ}}
\frac{\langle\alpha|\hat{j}_x|\beta\rangle\langle\beta|\hat{j}_y|\alpha\rangle-
\langle\alpha|\hat{j}_y|\beta\rangle\langle\beta|\hat{j}_x|\alpha\rangle}
{(E_\beta-E_\alpha)^2}.
$$ (eq-kubo-hall-523)

It is a matrix-element form of linear response: transport coefficients are fixed by virtual transitions between occupied and empty states.
:::

Evaluating this formula for $\nu$ completely filled Landau levels gives

$$
\sigma_{xy}=\nu\frac{e^2}{h},
$$

connecting perturbation theory to an exact quantized observable (details in ┬¦4.3.3).

### Summary

- One perturbative framework generates four experimentally central ideas: resonance, selection rules, decay rates, and transport coefficients.
- Harmonic driving emphasizes **coherent** transitions and Rabi control.
- Coupling to many final states gives **rates** (golden rule), hence lifetime and linewidth.
- Linear response uses the same first-order logic with $\delta\hat{H}=-\hat{\boldsymbol{j}}\cdot\delta\boldsymbol{A}$, yielding the Kubo formula.
- For integer-filled Landau levels, Kubo evaluation gives exact Hall quantization $\sigma_{xy}=\nu e^2/h$.
