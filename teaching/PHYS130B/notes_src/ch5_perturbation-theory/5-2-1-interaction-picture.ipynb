## Lecture Notes

When quantum systems interact with external fields that change in time—whether oscillating electromagnetic radiation or engineered control pulses—the full dynamics become intractable. The interaction picture offers a brilliant mathematical perspective: by choosing a reference frame that rotates with the unperturbed evolution, we strip away the rapid oscillations and expose only the slow, physically significant interactions. This reformulation transforms time-dependent problems from impossible to solvable.

### The Interaction Picture

#### Three Pictures of Quantum Mechanics

State evolution can be described in different pictures:

| Picture | Evolution of States | Evolution of Operators | Use |
|---------|---|---|---|
| **Schrödinger** | $i\hbar d|\psi_S\rangle/dt = H|\psi_S\rangle$ | Static: $O_S(t) = O(0)$ | General; states change, operators fixed |
| **Heisenberg** | Static: $|\psi_H\rangle = 	ext{const}$ | $O_H(t) = U^\dagger O_S U$ | Classical-like; dynamics in observables |
| **Interaction** | Driven by $V_I(t)$ | Driven by $H_0$ | Separates free propagation from interactions |

#### Definition and Transformations

In the **interaction picture**, we separate the Hamiltonian into two parts:
- **Free/unperturbed part** $H_0$: evolution is rapid but exactly solvable (oscillates at natural frequencies $\omega_n = E_n^{(0)}/\hbar$)
- **Interaction part** $V(t)$: slow compared to $H_0$ (weak coupling regime)

States and operators transform as:

$$

|\psi_I(t)\rangle = e^{iH_0 t/\hbar} |\psi_S(t)\rangle

$$ (eq-5a-psi-IP)

$$

O_I(t) = e^{iH_0 t/\hbar} O_S e^{-iH_0 t/\hbar}

$$ (eq-5a-O-IP)

The key insight: multiplying by $e^{iH_0 t/\hbar}$ **factors out the rapid oscillations** from $H_0$, leaving only the slow dynamics driven by $V(t)$.

#### Equation of Motion in the Interaction Picture

The interaction-picture state obeys a much simpler equation:

$$

\boxed{i\hbar \frac{d|\psi_I\rangle}{dt} = \lambda V_I(t) |\psi_I\rangle}

$$ (eq-5a-Schrod-IP)

where the **interaction-picture perturbation** is:

$$

\boxed{V_I(t) = e^{iH_0 t/\hbar} V(t) e^{-iH_0 t/\hbar}}

$$ (eq-5a-VI-def)

**Physical content**: $V_I(t)$ is the perturbation as viewed from a frame rotating with the free evolution. Rapid oscillations (at natural frequencies $\sim E_n^{(0)}/\hbar$) are removed, revealing only the slow dynamics.

:::{admonition} Derivation
:class: dropdown note

Start with the Schrödinger equation:

$$

i\hbar \frac{d|\psi_S\rangle}{dt} = H(t)|\psi_S\rangle = [H_0 + \lambda V(t)]|\psi_S\rangle

$$

Transform to the interaction picture using $|\psi_S\rangle = e^{-iH_0 t/\hbar}|\psi_I\rangle$:

$$

i\hbar \frac{d}{dt}\left(e^{-iH_0 t/\hbar}|\psi_I\rangle\right) = [H_0 + \lambda V(t)]e^{-iH_0 t/\hbar}|\psi_I\rangle

$$

Expand the left side using the product rule:

$$

i\hbar e^{-iH_0 t/\hbar}\left(-\frac{i}{\hbar}H_0|\psi_I\rangle + \frac{d|\psi_I\rangle}{dt}\right) = [H_0 + \lambda V(t)]e^{-iH_0 t/\hbar}|\psi_I\rangle

$$

Simplify:

$$

H_0 e^{-iH_0 t/\hbar}|\psi_I\rangle + i\hbar e^{-iH_0 t/\hbar}\frac{d|\psi_I\rangle}{dt} = H_0 e^{-iH_0 t/\hbar}|\psi_I\rangle + \lambda V(t)e^{-iH_0 t/\hbar}|\psi_I\rangle

$$

The $H_0$ terms cancel:

$$

i\hbar \frac{d|\psi_I\rangle}{dt} = \lambda e^{iH_0 t/\hbar} V(t) e^{-iH_0 t/\hbar}|\psi_I\rangle = \lambda V_I(t)|\psi_I\rangle

$$

:::

#### Eigenbasis Expansion

Expand $V_I(t)$ in the eigenbasis of $H_0$:

$$

V_I(t) = \sum_{n,m} e^{i\omega_{nm}t} V_{nm}(t) |n\rangle\langle m|

$$ (eq-5a-VI-eigenbasis)

where:

$$

\omega_{nm} = \frac{E_n^{(0)} - E_m^{(0)}}{\hbar}

$$ (eq-5a-transition-freq)

The oscillatory factor $e^{i\omega_{nm}t}$ encodes **resonance**: 
- When $\omega_{nm} \approx 0$ (initial and final states have nearly equal energy): coupling is strong, transitions occur.
- When $|\omega_{nm}|$ is large (states far apart in energy): rapid oscillations suppress the coupling.

### The Evolution Operator and Dyson Series

#### Formal Solution

Define the **interaction-picture evolution operator** $U_I(t)$ such that:

$$

|\psi_I(t)\rangle = U_I(t) |\psi_I(0)\rangle

$$ (eq-5a-UI-def)

This operator has the formal solution:

$$

\boxed{U_I(t) = T \exp\left(-\frac{i\lambda}{\hbar} \int_0^t dt' V_I(t')\right)}

$$ (eq-5a-time-ord-exp)

This is the **time-ordered exponential**. Expanding it gives the **Dyson series**:

$$

\boxed{U_I(t) = \mathbb{I} + \frac{\lambda}{i\hbar} \int_0^t dt_1 V_I(t_1) + \left(\frac{\lambda}{i\hbar}\right)^2 \int_0^t dt_1 \int_0^{t_1} dt_2 V_I(t_1) V_I(t_2) + \ldots}

$$ (eq-5a-dyson-expanded)

**Physical meaning**:
- **Zeroth order** ($\lambda^0$): No interaction, system stays in its initial state.
- **First order** ($\lambda^1$): One interaction event at time $t_1$.
- **Second order** ($\lambda^2$): Two interaction events at times $t_1$ and $t_2$, with $t_1 > t_2$ (time-ordered).
- **Higher orders**: Multiple interaction events, always ordered in time ($t_1 > t_2 > t_3 > \ldots$).

Each term represents a different "scattering" process or number of intermediate transitions.

:::{admonition} Derivation
:class: dropdown note

From the equation of motion:

$$

i\hbar \frac{\partial U_I}{\partial t} = \lambda V_I(t) U_I(t), \quad U_I(0) = \mathbb{I}

$$

Integrate formally:

$$

U_I(t) = \mathbb{I} + \frac{1}{i\hbar} \lambda \int_0^t dt_1 V_I(t_1) U_I(t_1)

$$

This is a **Volterra integral equation**. Solve by iterating: substitute the expression for $U_I$ back into the integral to obtain higher orders:

**First order**:

$$

U_I^{(1)}(t) = \frac{\lambda}{i\hbar} \int_0^t dt_1 V_I(t_1)

$$

**Second order**:

$$

U_I^{(2)}(t) = \left(\frac{\lambda}{i\hbar}\right)^2 \int_0^t dt_1 V_I(t_1) \int_0^{t_1} dt_2 V_I(t_2)

$$

The integral limits automatically enforce time ordering: when we substitute $U_I(t_1)$ into the outer integral, the inner integration runs from $0$ to $t_1$, ensuring $t_1 > t_2$. This pattern continues for all higher orders.

Summing all orders:

$$

U_I(t) = \sum_{n=0}^\infty \left(\frac{\lambda}{i\hbar}\right)^n \int_0^t dt_1 \int_0^{t_1} dt_2 \cdots \int_0^{t_{n-1}} dt_n \, V_I(t_1) V_I(t_2) \cdots V_I(t_n)

$$

This is precisely the expansion of the time-ordered exponential.

:::

#### Time Ordering and Causality

The integral limits $t > t_1 > t_2 > \ldots$ ensure **causal time ordering**: interactions are processed in sequence, with later times to the left.

**Why this matters**: In general, $[V_I(t_1), V_I(t_2)] \neq 0$ (operators do not commute). The order in which interactions occur is physical and affects the result. Time ordering enforces the natural causal sequence.

:::{admonition} Time-ordering operator definition
:class: note

For any two operators $A(t_1)$ and $B(t_2)$ with $t_1 > t_2$, the time-ordering operator $T$ places later times to the left:

$$

T[A(t_1) B(t_2)] = A(t_1) B(t_2) \quad (t_1 > t_2)

$$

$$

T[B(t_2) A(t_1)] = A(t_1) B(t_2) \quad (t_1 > t_2)

$$

The **time-ordered exponential** $T \exp(X)$ automatically generates all orders of an expansion with time-ordered integration limits.

:::

#### Full Schrödinger-Picture Evolution

The **Schrödinger-picture evolution operator** $U_S(t)$ factorizes as:

$$

\boxed{U_S(t) = e^{-iH_0 t/\hbar} U_I(t)}

$$ (eq-5a-US-factored)

This decomposition is remarkably powerful: 
- The first factor $e^{-iH_0 t/\hbar}$ handles the fast, high-frequency $H_0$ oscillations.
- The second factor $U_I(t)$ (via the Dyson series) captures the slow interactions.

Together, they build up the full interacting dynamics without having to solve the complete Hamiltonian.

### Perturbative Expansion and Transition Amplitudes

#### Order-by-Order Structure

Expand the evolution operator in powers of $\lambda$:

$$

U_I(t) = U_I^{(0)}(t) + \lambda U_I^{(1)}(t) + \lambda^2 U_I^{(2)}(t) + \ldots

$$ (eq-5a-UI-series)

**Zeroth order** ($\lambda = 0$): 

$$

U_I^{(0)}(t) = \mathbb{I} \quad \text{(no interaction)}

$$

**First order** ($\lambda^1$): One interaction at time $t_1$.

$$

U_I^{(1)}(t) = -\frac{i}{\hbar} \int_0^t dt_1 V_I(t_1)

$$

**Second order** ($\lambda^2$): Two interactions at times $t_1, t_2$ with $t_1 > t_2$.

$$

U_I^{(2)}(t) = -\frac{1}{\hbar^2} \int_0^t dt_1 \int_0^{t_1} dt_2 V_I(t_1) V_I(t_2)

$$

Higher orders follow the pattern: more interaction times, more nested integrals, time-ordered products.

#### Transition Amplitudes

The transition amplitude from state $|i\rangle$ to $|f\rangle$ is:

$$

\mathcal{A}_{i\to f}(t) = \langle f | U_S(t) | i \rangle

$$ (eq-5a-amp-def)

To first order in $\lambda$:

$$

\mathcal{A}_{i\to f}^{(0)}(t) = \delta_{fi} e^{-iE_f^{(0)} t/\hbar}

$$

$$

\mathcal{A}_{i\to f}^{(1)}(t) = -\frac{i}{\hbar} e^{-iE_f^{(0)} t/\hbar} \int_0^t dt_1 e^{-i\omega_{fi}t_1} \langle f | V(t_1) | i \rangle

$$ (eq-5a-amp-1st)

where $\omega_{fi} = (E_f^{(0)} - E_i^{(0)})/\hbar$ is the transition frequency.

### Physical Interpretation

#### Resonance and Detuning

The oscillatory factor $e^{-i\omega_{fi}t_1}$ in the transition amplitude has frequency $\omega_{fi}$, determined by the energy difference between initial and final states.

**Resonant condition** ($\omega_{fi} \approx 0$): The integrand oscillates slowly or not at all; the integral accumulates, giving a large (growing) amplitude. This is a strong resonant transition.

**Off-resonant condition** ($|\omega_{fi}| \gg 1/T$ for evolution time $T$): Rapid oscillations cause cancellation; amplitude is suppressed. Weak off-resonant transitions.

This is the origin of the **rotating wave approximation**: only near-resonant transitions (with $|\omega_{fi}| \lesssim 1/T$) contribute significantly.

#### Virtual Processes

Higher-order terms in the Dyson series correspond to **virtual intermediate states**:

- **First order**: Direct $i \to f$ transition.
- **Second order**: $i \to n \to f$ (via intermediate state $n$), where state $n$ is visited transiently.
- **Higher orders**: Chains of intermediate transitions.

The time integrals in the Dyson series encode **energy conservation on average**: the system is not required to have fixed energy during intermediate steps (virtual processes), but the overall energy is conserved when we integrate over all interaction times.

### Summary

- **Interaction picture**: Separates fast (solved by $e^{-iH_0 t/\hbar}$) and slow (perturbative $V_I(t)$) dynamics.
- **Interaction-picture perturbation**: $V_I(t) = e^{iH_0 t/\hbar} V(t) e^{-iH_0 t/\hbar}$, with oscillatory terms $e^{i\omega_{nm}t}$ encoding resonances.
- **Evolution operator**: $U_I(t) = T \exp(-i \int V_I dt / \hbar)$ (time-ordered exponential).
- **Dyson series**: Systematic expansion showing interaction as sequence of events at ordered times.
- **Full evolution**: $U_S(t) = e^{-iH_0 t/\hbar} U_I(t)$ factorizes free propagation and interactions.
- **Transition amplitudes**: Depend on oscillatory factors $e^{i\omega_{fi}t}$; resonance ($\omega_{fi} \approx 0$) enhances transitions.
- **Causality**: Time ordering ensures that interactions occur in the correct physical sequence.
- **Virtual processes**: Higher-order terms represent intermediate state transitions with average energy conservation.
