## Lecture Notes

### Overview

The goal of time-dependent perturbation theory is parallel to the toy-model logic in 5.1:

1. start from something exactly solvable,
2. isolate a small perturbation,
3. expand observables/evolution order by order.

For dynamics, the exactly solvable object is the free evolution from $\hat{H}_0$. The interaction picture is the frame that cleanly separates this free part from the perturbation.

### Problem Setup

Assume

$$
\hat{H}(t)=\hat{H}_0+\hat{V}(t),
$$

where $\hat{H}_0$ is exactly solvable and $\hat{V}(t)$ is weak (or treated perturbatively).

If

$$
\hat{U}_0(t)=\mathrm{e}^{-\mathrm{i}\hat{H}_0 t},
$$

then the strategy is: move to the co-moving frame generated by $\hat{U}_0(t)$, so perturbation theory is performed on $\hat{V}(t)$ rather than on the full fast dynamics.

### Interaction Picture: Definition and Equation

Define

$$
\vert\psi(t)\rangle_{\mathcal{I}}=\hat{U}_0^{\dagger}(t)\vert\psi(t)\rangle_{\mathcal{S}},
$$

and

$$
\hat{V}_{\mathcal{I}}(t)=\hat{U}_0^{\dagger}(t)\hat{V}(t)\hat{U}_0(t).
$$

Then the state equation becomes

$$
\mathrm{i}\partial_t\vert\psi(t)\rangle_{\mathcal{I}}=\hat{V}_{\mathcal{I}}(t)\vert\psi(t)\rangle_{\mathcal{I}}.
$$ (eq-5-2-1-ip-eom)

:::{admonition} Derivation (one line of logic)
:class: dropdown info

Insert
$\vert\psi\rangle_{\mathcal{S}}=\hat{U}_0\vert\psi\rangle_{\mathcal{I}}$
into
$\mathrm{i}\partial_t\vert\psi\rangle_{\mathcal{S}}=(\hat{H}_0+\hat{V})\vert\psi\rangle_{\mathcal{S}}$,
apply product rule, and cancel the $\hat{H}_0$ terms. What remains is exactly Eq. {eq}`eq-5-2-1-ip-eom`.
:::

In the $\hat{H}_0$ eigenbasis, matrix elements are

$$
\langle m\vert\hat{V}_{\mathcal{I}}(t)\vert n\rangle = V_{mn}(t)\,\mathrm{e}^{\mathrm{i}(E_m-E_n)t}.
$$

This makes the physical filter explicit: rapidly oscillating terms average out, while slowly varying (near-resonant) terms accumulate.

### Evolution Operator and Dyson Series

Define interaction-picture evolution by

$$
\vert\psi(t)\rangle_{\mathcal{I}}=\hat{U}_{\mathcal{I}}(t)\vert\psi(0)\rangle_{\mathcal{I}},
$$

so

$$
\mathrm{i}\partial_t\hat{U}_{\mathcal{I}}(t)=\hat{V}_{\mathcal{I}}(t)\hat{U}_{\mathcal{I}}(t),\qquad
\hat{U}_{\mathcal{I}}(0)=\mathbf{1}.
$$

Integrating once gives the exact integral equation

$$
\hat{U}_{\mathcal{I}}(t)=\mathbf{1}-\mathrm{i}\int_0^t\mathrm{d}t'\,\hat{V}_{\mathcal{I}}(t')\hat{U}_{\mathcal{I}}(t').
$$

Iterating this equation generates the Dyson series:

$$
\hat{U}_{\mathcal{I}}(t)=\sum_{k=0}^{\infty}(-\mathrm{i})^k
\int_0^t\mathrm{d}t_k\int_0^{t_k}\mathrm{d}t_{k-1}\cdots\int_0^{t_2}\mathrm{d}t_1\,
\hat{V}_{\mathcal{I}}(t_k)\hat{V}_{\mathcal{I}}(t_{k-1})\cdots\hat{V}_{\mathcal{I}}(t_1).
$$ (eq-5-2-1-dyson)

The nested bounds enforce time ordering automatically: earlier interactions act on the right, later ones on the left.

### Physical Reading of the Expansion

- **0th order:** no interaction, pure free evolution.
- **1st order:** one interaction event (single transition channel).
- **2nd and higher:** sequences of virtual intermediate processes.

This is the dynamic analogue of 5.1 non-degenerate perturbation theory: same order-by-order logic, but now expanded in interaction events across time.

### Minimal Two-Level Example (Resonant Drive)

For a qubit with level spacing $\omega_0$ and drive near resonance,

$$
\hat{V}(t)\propto \cos(\omega t)\,\hat{X},
$$

$\hat{V}_{\mathcal{I}}(t)$ splits into rotating pieces. Near $\omega\approx\omega_0$, the slow piece dominates while the fast counter-rotating piece averages out (rotating-wave approximation), producing coherent Rabi oscillations.

### Summary

- Interaction picture is a **frame choice**: remove known free evolution, isolate perturbative dynamics.
- The equation of motion is governed solely by $\hat{V}_{\mathcal{I}}(t)$, making perturbative structure explicit.
- Dyson series comes from iterating an exact integral equation; each order counts interaction events in time order.
- Resonance appears naturally through phase factors in $\hat{V}_{\mathcal{I}}(t)$: slow phases drive real transitions, fast phases average out.

**2. Transition frequencies.** The unperturbed Hamiltonian $\hat{H}_0$ has eigenstates $\{\vert n\rangle\}$ with energies $\{E_n^{(0)}\}$.

(a) Show that in the interaction picture, the matrix elements of $V_I(t)$ are $\langle m\vert V_I(t)\vert n\rangle = V_{mn}(t)\,\mathrm{e}^{\mathrm{i}\omega_{mn}t}$ where $\omega_{mn} = (E_m^{(0)} - E_n^{(0)})/\hbar$.

(b) Interpret the oscillatory factors: why do matrix elements between nearly degenerate states ($\omega_{mn} \approx 0$) evolve slowly, while those between widely separated states oscillate rapidly?

**3. Dyson series.** The interaction-picture evolution operator satisfies $\mathrm{i}\hbar\partial_t U_I = V_I(t)U_I$ with $U_I(0) = \hat{I}$.

(a) Integrate formally to obtain $U_I(t) = \hat{I} - \frac{\mathrm{i}}{\hbar}\int_0^t V_I(t_1)\,U_I(t_1)\,\mathrm{d}t_1$.

(b) Iterate once to write the Dyson series to second order. Verify that the integration limits $t > t_1 > t_2$ enforce time ordering.

(c) Explain why time ordering is necessary: what goes wrong if $V_I(t_1)$ and $V_I(t_2)$ do not commute and we ignore their ordering?

**4. Rabi oscillations.** A two-level atom with $\vert 0\rangle$ (energy $0$) and $\vert 1\rangle$ (energy $\hbar\omega_0$) is driven by $V(t) = \hbar\Omega\hat{X}\cos(\omega_0 t)$.

(a) Transform to the interaction picture and decompose $V_I(t)$ into rotating and counter-rotating terms. Identify which term oscillates slowly near resonance.

(b) In the rotating-wave approximation (dropping the fast term), write the coupled equations for $a_0(t)$ and $a_1(t)$ where $\vert\psi_I\rangle = a_0\vert 0\rangle + a_1\vert 1\rangle$.

(c) Solve with initial condition $\vert\psi_I(0)\rangle = \vert 0\rangle$ and show that $\hat{P}_1(t) = \sin^2(\Omega t/2)$. What drive duration gives a $\pi$-pulse (complete inversion)?

**5. First-order transition.** A system starts in eigenstate $\vert n\rangle$ of $\hat{H}_0$ and evolves under a constant perturbation $V$ turned on at $t = 0$.

(a) Using the first-order Dyson series, show that the transition amplitude to state $\vert m\rangle$ is $a_m(t) = -\frac{\mathrm{i}}{\hbar}V_{mn}\frac{\mathrm{e}^{\mathrm{i}\omega_{mn}t} - 1}{\mathrm{i}\omega_{mn}}$.

(b) Compute the transition probability $P_{n\to m}(t) = \vert a_m(t)\vert^2$ and show it involves a sinc-like function peaked at $\omega_{mn} = 0$.
