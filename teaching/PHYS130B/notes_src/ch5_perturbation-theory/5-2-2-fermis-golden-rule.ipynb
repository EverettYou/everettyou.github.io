# 5.2.2 Fermi's Golden Rule

:::{admonition} Prompts
:class: tip

- In time-dependent perturbation theory, how does a finite-time resonance window evolve into energy-conserving transitions in the long-time limit?
- What is Fermi's golden rule physically, and why must the rate depend on both coupling strength and density of available final states?
- How do symmetry and operator structure produce selection rules, and why can some channels be strongly suppressed at leading order?
- Why are lifetime and linewidth two sides of the same physics for unstable excited states?

## Lecture Notes

### Overview

This lecture turns first-order time-dependent perturbation theory into a measurable prediction for transition rates. The core flow is:

1. Start from the transition amplitude in the interaction picture.
2. For a monochromatic drive, obtain the sinc-squared line shape.
3. Take the long-time limit to extract a constant rate (Fermi's golden rule).
4. Interpret the result through matrix elements, density of states, and symmetry-based selection rules.

### From Amplitude to Probability

For an initial eigenstate $\\vert i\\rangle$ of $\hat{H}_0$, the first-order transition probability to $\\vert f\\rangle$ is

$$
P_{i\to f}^{(1)}(t)=\frac{1}{\hbar^2}\left|\int_0^t\mathrm{d}t_1\,\mathrm{e}^{\mathrm{i}\omega_{fi}t_1}V_{fi}(t_1)\right|^2,
$$ (eq-5b-P1st-general)

where $\omega_{fi}=(E_f^{(0)}-E_i^{(0)})/\hbar$ and $V_{fi}(t)=\langle f\vert \hat V(t)\vert i\rangle$.

For a drive switched on at $t=0$ with $\hat V(t)=\hat V\,\mathrm{e}^{-\mathrm{i}\omega t}$ ($t>0$), the time integral gives

$$
P_{i\to f}^{(1)}(t)=\frac{|V_{fi}|^2}{\hbar^2}\left[\frac{\sin\left((\omega_{fi}-\omega)t/2\right)}{(\omega_{fi}-\omega)/2}\right]^2.
$$ (eq-5b-sinc-driven)

This sinc-squared factor is a finite-time resonance window.

:::{admonition} Physical meaning of the sinc window
:class: note

- Peak at resonance: $\omega_{fi}=\omega$.
- Peak height scales as $t^2$.
- Peak width scales as $1/t$.
- Area scales as $t$, which is why a constant transition rate emerges at long times.
:::

### Long-Time Limit and Golden Rule

For a continuum of final states,

$$
P_i(t)=\int \mathrm{d}E_f\,\rho(E_f)\,P_{i\to f}^{(1)}(t).
$$

Using

$$
\lim_{t\to\infty}\frac{1}{t}\left[\frac{\sin(xt/2)}{x/2}\right]^2 = 2\pi\,\delta(x),
$$

the long-time transition rate is

:::{admonition} Fermi's Golden Rule
:class: important

$$
\Gamma_{i\to f}=\frac{2\pi}{\hbar}|\langle f\vert\hat V\vert i\rangle|^2\,\rho(E_f)\Big|_{E_f=E_i+\hbar\omega}.
$$ (eq-5b-fgr-standard)

For a static perturbation ($\omega=0$), this reduces to $E_f=E_i$.
:::

:::{admonition} Why this formula is powerful
:class: tip

- **Dynamics** enters through $|\langle f\vert\hat V\vert i\rangle|^2$.
- **Kinematics** enters through $\rho(E)$.
- **Conservation law** enters through the resonance condition selected by the delta function.
:::

### Validity Regime

:::{admonition} When to trust the golden rule
:class: caution

- Coupling is weak (first-order perturbation remains controlled).
- Observation time is long compared with oscillation scales, but still short compared with strong-depletion times.
- Final states form a dense spectrum so sums can be replaced by an integral over $\rho(E)$.

Outside this regime (strong drive, sparse spectrum, coherent Rabi oscillations), the golden-rule picture fails.
:::

### Selection Rules and Linewidth

The matrix element can vanish by symmetry, producing selection rules. For electric dipole transitions,

- $\Delta \ell=\pm 1$,
- $\Delta m_\ell=0,\pm 1$ (polarization dependent),
- $\Delta S=0$ (at leading order).

Thus, "forbidden" transitions are strongly suppressed at leading order and often proceed through higher multipoles.

If an excited state decays with rate $\Gamma$, it has finite lifetime $\tau=1/\Gamma$. The corresponding spectral line has natural width

$$
\Delta E\sim \hbar\Gamma,
$$

connecting decay dynamics to measurable spectroscopy linewidth.

:::{admonition} Discussion: short-time vs long-time behavior
:class: dropdown tip

At short times, transition probability grows quadratically ($P\propto t^2$). At long times in a continuum, it grows linearly ($P\approx \Gamma t$). How does this crossover clarify when the golden rule is appropriate, and when coherent dynamics (e.g., Zeno physics or Rabi oscillations) is the better description?
:::

### Summary

- First-order perturbation gives a sinc-squared resonance profile in time-dependent transitions.
- In the long-time continuum limit, sinc-squared becomes a delta function and yields Fermi's golden rule.
- The rate is controlled by matrix element strength, density of final states, and resonance/energy conservation.
- Selection rules explain why some channels are absent at leading order, while finite lifetime implies finite linewidth.
