## Lecture Notes

### Overview

Sec. 5.1.2 assumed isolated levels and produced scalar corrections. This subsection handles the complementary case: when unperturbed levels are degenerate, perturbation theory must be organized as a block problem inside each degenerate manifold.

The practical workflow is: identify the degenerate manifold, diagonalize in-manifold matrix elements at first order, then add virtual-coupling corrections from outside manifolds at second order.

### Why non-degenerate formulas fail

For $\hat{H}(\lambda)=\hat{H}_0+\lambda\hat{V}$, suppose $\hat{H}_0$ has a $d_n$-fold degenerate eigenspace at energy $E_n$, spanned by $|n,\alpha\rangle$ ($\alpha=1,\dots,d_n$):

$$
\hat{H}_0|n,\alpha\rangle=E_n|n,\alpha\rangle.
$$

A quick wrong attempt is to reuse the non-degenerate structure directly:

$$
|n,\alpha\rangle^{(1)}\sim\sum_{(m,\beta)\neq(n,\alpha)}|m,\beta\rangle\frac{(\cdots)}{E_n-E_m}.
$$

Then terms with $m=n$ but $\beta\neq\alpha$ give

$$
E_n-E_m=E_n-E_n=0,
$$

so the denominator diverges.

:::{admonition} Core Idea
:class: important

The proper strategy is **divide and conquer**:

- reorganize the Hilbert space into degenerate-manifold blocks,
- construct the effective Hamiltonian for each block,
- diagonalize each block,
- if residual degeneracy remains, continue to the next order/block level.

The central goal of degenerate perturbation theory is to construct the **effective Hamiltonian** within each degenerate subspace.
:::

### Problem Setup

:::{admonition} Degenerate Perturbation Problem
:class: important

Consider $\hat H(\lambda)=\hat H_0+\lambda\hat V$, and work in a basis $|n,\alpha\rangle$ where $n$ labels the manifold and $\alpha=1,\dots,d_n$ labels orthonormal states inside that manifold:

$$
\hat H_0=\sum_n\sum_\alpha E_n\,|n,\alpha\rangle\langle n,\alpha|.
$$

$$
\hat V=\sum_{m,\alpha}\sum_{n,\beta}|m,\alpha\rangle V_{m\alpha,n\beta}\langle n,\beta|.
$$

Here $V_{m\alpha,n\beta}=\langle m,\alpha|\hat V|n,\beta\rangle$, and states inside one degenerate manifold can mix.
:::

The corresponding eigenvalue equation is $\hat H(\lambda)|n,\beta(\lambda)\rangle=\sum_\alpha E_{n,\alpha\beta}(\lambda)|n,\alpha(\lambda)\rangle$, and our objective is to construct $E_{n,\alpha\beta}(\lambda)$ and $|n,\alpha(\lambda)\rangle$ order by order in $\lambda$.

### Hellmann-Feynman Identities (Degenerate Form)

Differentiate the matrix eigenproblem and project onto $\langle m,\gamma|$ at $\lambda=0$:

$$
\langle m,\gamma|\partial_\lambda\hat H|n,\beta\rangle
=\partial_\lambda E_{n,\gamma\beta}\,\delta_{mn}
+(E_n-E_m)\langle m,\gamma|\partial_\lambda n,\beta\rangle.
$$

:::{admonition} Hellmann-Feynman identities (degenerate)
:class: important

- **1st Hellmann-Feynman Identity** (energy matrix derivative, within manifold):

$$
\partial_\lambda E_{n,\alpha\beta}=V_{n\alpha,n\beta}.
$$

- **2nd Hellmann-Feynman Identity** (state-mixing derivative, across manifolds):

$$
\langle m,\alpha|\partial_\lambda n,\beta\rangle
=\frac{V_{m\alpha,n\beta}}{E_n-E_m}
\text{ for }m\neq n.
$$
:::

:::{admonition} Derivation: degenerate Hellmann-Feynman identities
:class: dropdown information

**Step 1: Differentiate the matrix eigenvalue equation**

Start from

$$
\hat H(\lambda)|n,\beta(\lambda)\rangle
=\sum_\alpha E_{n,\alpha\beta}(\lambda)\,|n,\alpha(\lambda)\rangle.
$$

Differentiate with respect to $\lambda$:

$$
\begin{aligned}
\partial_\lambda\hat H\,|n,\beta\rangle+\hat H\,|\partial_\lambda n,\beta\rangle
=\sum_\alpha\Big[
(\partial_\lambda E_{n,\alpha\beta})|n,\alpha\rangle
+E_{n,\alpha\beta}|\partial_\lambda n,\alpha\rangle
\Big].
\end{aligned}
$$

**Step 2: Project with $\langle m,\gamma|$ and evaluate at $\lambda=0$**

Left-multiply by $\langle m,\gamma|$:

$$
\begin{aligned}
\langle m,\gamma|\partial_\lambda\hat H|n,\beta\rangle
+\langle m,\gamma|\hat H|\partial_\lambda n,\beta\rangle
=\sum_\alpha\Big[
\partial_\lambda E_{n,\alpha\beta}\,\langle m,\gamma|n,\alpha\rangle
+E_{n,\alpha\beta}\,\langle m,\gamma|\partial_\lambda n,\alpha\rangle
\Big].
\end{aligned}
$$

Evaluate at $\lambda=0$:

$$
V_{m\gamma,n\beta}+E_m\langle m,\gamma|\partial_\lambda n,\beta\rangle
=\delta_{mn}\,\partial_\lambda E_{n,\gamma\beta}+E_n\langle m,\gamma|\partial_\lambda n,\beta\rangle.
$$

So

$$
V_{m\gamma,n\beta}
=\delta_{mn}\,\partial_\lambda E_{n,\gamma\beta}
+(E_n-E_m)\langle m,\gamma|\partial_\lambda n,\beta\rangle.
$$

**Step 3: Split into $m=n$ and $m\neq n$**

If $m=n$,

$$
\partial_\lambda E_{n,\alpha\beta}=V_{n\alpha,n\beta}.
$$

If $m\neq n$,

$$
\langle m,\alpha|\partial_\lambda n,\beta\rangle
=\frac{V_{m\alpha,n\beta}}{E_n-E_m}.
$$

These are the two degenerate Hellmann-Feynman identities used below.
:::

### Energy Corrections

Use the Taylor expansion of the in-manifold energy matrix:

:::{admonition} Energy matrix expansion up to second order
:class: important

Using degenerate Hellmann-Feynman identities, the energy matrix correction is given by:

$$
\begin{split}
E_{n,\alpha\beta}(\lambda)
&=E_n\delta_{\alpha\beta}
+\lambda\,\partial_\lambda E_{n,\alpha\beta}
+\frac{\lambda^2}{2}\,\partial_\lambda^2 E_{n,\alpha\beta}
+O(\lambda^3)\\
&=E_n\delta_{\alpha\beta}
+\lambda V_{n\alpha,n\beta}
+\lambda^2\sum_{m\neq n}\sum_\gamma
\frac{V_{n\alpha,m\gamma}V_{m\gamma,n\beta}}{E_n-E_m}
+O(\lambda^3).
\end{split}
$$
:::

:::{admonition} Derivation: first-order energy matrix derivative
:class: dropdown information

Set $m=n$ in the projected derivative equation:

$$
\langle n,\alpha|\partial_\lambda \hat H|n,\beta\rangle
=\partial_\lambda E_{n,\alpha\beta}.
$$

With $\partial_\lambda\hat H=\hat V$, this gives

$$
\partial_\lambda E_{n,\alpha\beta}=V_{n\alpha,n\beta}.
$$
:::

:::{admonition} Derivation: second-order energy matrix derivative
:class: dropdown information

Differentiate
$\partial_\lambda E_{n,\alpha\beta}=\langle n,\alpha|\hat V|n,\beta\rangle$ once more:

$$
\partial_\lambda^2 E_{n,\alpha\beta}
=\langle\partial_\lambda n,\alpha|\hat V|n,\beta\rangle
+\langle n,\alpha|\hat V|\partial_\lambda n,\beta\rangle.
$$

Insert a complete basis explicitly:

$$
\partial_\lambda^2 E_{n,\alpha\beta}
=\sum_{m,\gamma}\Big(\langle\partial_\lambda n,\alpha|m,\gamma\rangle V_{m\gamma,n\beta}
+V_{n\alpha,m\gamma}\langle m,\gamma|\partial_\lambda n,\beta\rangle\Big).
$$

Now split into $m\neq n$ and $m=n$ pieces. For $m\neq n$, use the second Hellmann-Feynman identity:

$$
\langle m,\gamma|\partial_\lambda n,\beta\rangle
=\frac{V_{m\gamma,n\beta}}{E_n-E_m},
$$

and its conjugate relation, giving

$$
2\sum_{m\neq n}\sum_\gamma
\frac{V_{n\alpha,m\gamma}V_{m\gamma,n\beta}}{E_n-E_m}.
$$

For $m=n$, terms involve in-manifold overlaps
$A_{\gamma\beta}=\langle n,\gamma|\partial_\lambda n,\beta\rangle$.
These are gauge/basis-choice dependent within the degenerate manifold and can be set to zero by parallel-transport gauge after fixing the first-order in-manifold basis.

Therefore

$$
\partial_\lambda^2 E_{n,\alpha\beta}
=2\sum_{m\neq n}\sum_\gamma
\frac{V_{n\alpha,m\gamma}V_{m\gamma,n\beta}}{E_n-E_m}.
$$
:::

### State Corrections

The corrected basis vectors are also expanded in $\lambda$.

:::{admonition} State expansion and first-order mixing
:class: important

Using the second Hellmann-Feynman identity, the state correction is given by:

$$
\begin{split}
|n,\alpha(\lambda)\rangle
&=|n,\alpha\rangle
+\lambda|\partial_\lambda n,\alpha\rangle
+O(\lambda^2)\\
&=|n,\alpha\rangle
+\lambda\sum_{m\neq n}\sum_\beta
|m,\beta\rangle\frac{V_{m\beta,n\alpha}}{E_n-E_m}
+O(\lambda^2).
\end{split}
$$
:::

:::{admonition} Derivation: first-order state derivative
:class: dropdown information

Expand in the unperturbed basis and separate the two parts:

$$
|\partial_\lambda n,\alpha\rangle
=\sum_{\beta}|n,\beta\rangle\langle n,\beta|\partial_\lambda n,\alpha\rangle
+\sum_{m\neq n,\beta}|m,\beta\rangle\langle m,\beta|\partial_\lambda n,\alpha\rangle.
$$

For $m\neq n$, apply

$$
\langle m,\beta|\partial_\lambda n,\alpha\rangle
=\frac{V_{m\beta,n\alpha}}{E_n-E_m}.
$$

For $m=n$, define in-manifold overlaps
$A_{\beta\alpha}=\langle n,\beta|\partial_\lambda n,\alpha\rangle$.
From orthonormality, $A$ is anti-Hermitian, and by basis/gauge choice inside the degenerate manifold (after first-order diagonalization) one may set $A=0$.

Under that gauge,

$$
|\partial_\lambda n,\alpha\rangle
=\sum_{m\neq n,\beta}|m,\beta\rangle\frac{V_{m\beta,n\alpha}}{E_n-E_m}.
$$
:::

:::{admonition} Note: Operator Form
:class: dropdown note

The same result can be written with projectors
$\hat{P}_n=\sum_\alpha |n,\alpha\rangle\langle n,\alpha|$ and
$\hat{Q}_n=\hat I-\hat P_n$:

$$
\hat H^{\text{eff}}_n
=E_n\hat I_d
+\lambda\hat P_n\hat V\hat P_n
+\lambda^2\hat P_n\hat V\hat Q_n
\frac{1}{E_n-\hat Q_n\hat H_0\hat Q_n}
\hat Q_n\hat V\hat P_n
+O(\lambda^3).
$$

This is exactly the same content as the component formulas above:

- $\hat P_n\hat V\hat P_n \leftrightarrow V_{n\alpha,n\beta}$ (first-order block),
- second-order projector term
  $\leftrightarrow \sum_{m\neq n,\gamma}V_{n\alpha,m\gamma}V_{m\gamma,n\beta}/(E_n-E_m)$.

Use component form for calculations; use operator form as a compact summary.
:::


:::{admonition} Example: Three-level bright/dark mechanism
:class: dropdown example

**Problem.**
Consider

$$
\hat H_0=\Delta|3\rangle\langle3|,
$$

$$
\hat V=\lambda_1|1\rangle\langle3|+\lambda_2|2\rangle\langle3|+\mathrm{h.c.},
$$

with $|\lambda_{1,2}|\ll\Delta$.
The degenerate manifold is $\{|1\rangle,|2\rangle\}$ at energy $0$.

**Step 1. First-order block check.**
Inside the manifold,

$$
(H^{\text{eff}})_{ij}^{(1)}=\langle i|\hat V|j\rangle=0,\qquad i,j\in\{1,2\},
$$

so first-order splitting is absent.

**Step 2. Build second-order block element-by-element.**
Because the only outside state is $|3\rangle$ with energy $\Delta$,

$$
\begin{split}
(H^{\text{eff}})_{ij}^{(2)}
&=\sum_{m\notin\{1,2\}}\frac{\langle i|\hat V|m\rangle\langle m|\hat V|j\rangle}{0-E_m}\\
&=-\frac{1}{\Delta}\langle i|\hat V|3\rangle\langle3|\hat V|j\rangle.
\end{split}
$$

Now evaluate four entries explicitly:

$$
(H^{\text{eff}})_{11}^{(2)}=-\frac{|\lambda_1|^2}{\Delta},
\qquad
(H^{\text{eff}})_{22}^{(2)}=-\frac{|\lambda_2|^2}{\Delta},
$$

$$
(H^{\text{eff}})_{12}^{(2)}=-\frac{\lambda_1\lambda_2^*}{\Delta},
\qquad
(H^{\text{eff}})_{21}^{(2)}=-\frac{\lambda_2\lambda_1^*}{\Delta}.
$$

Hence

$$
\hat H^{\text{eff}}=-\frac{1}{\Delta}
\begin{pmatrix}
|\lambda_1|^2 & \lambda_1\lambda_2^*\\
\lambda_2\lambda_1^* & |\lambda_2|^2
\end{pmatrix}
=-\frac{1}{\Delta}|\lambda\rangle\langle\lambda|,
$$

with $|\lambda\rangle=\lambda_1|1\rangle+\lambda_2|2\rangle$.

**Step 3. Diagonalize and identify bright/dark states.**
This is a rank-1 projector form, so eigenvalues are immediate:

$$
E_B=-\frac{|\lambda_1|^2+|\lambda_2|^2}{\Delta},
\qquad
E_D=0.
$$

A normalized bright state is

$$
|B\rangle=
\frac{\lambda_1|1\rangle+\lambda_2|2\rangle}{\sqrt{|\lambda_1|^2+|\lambda_2|^2}},
$$

and a normalized dark state orthogonal to $|B\rangle$ is

$$
|D\rangle=
\frac{\lambda_2^*|1\rangle-\lambda_1^*|2\rangle}{\sqrt{|\lambda_1|^2+|\lambda_2|^2}}.
$$

**Step 4. Exact zero-mode check (not only perturbative).**

$$
\hat H_0|D\rangle=0,
\qquad
\hat V|D\rangle=0,
$$

so $(\hat H_0+\hat V)|D\rangle=0$ exactly.

**Interpretation.**
First-order splitting is absent, but second-order virtual coupling generates an effective interaction with one bright mode and one protected dark mode.
:::


:::{admonition} Example: Spin-1 splitting from a degenerate manifold
:class: dropdown example

**Problem.**
Consider a spin-1 system in basis $\{|+1\rangle,|0\rangle,|-1\rangle\}$ with

$$
\hat H(\lambda)=\hat H_0+\lambda\hat V,
\qquad
\hat H_0=(\hat S^z)^2,
\qquad
\hat V=\hat S^x+\hat S^z.
$$

The operator matrices are

$$
\hat S^x\simeq \frac{1}{\sqrt2}
\begin{pmatrix}
0&1&0\\
1&0&1\\
0&1&0
\end{pmatrix},
\qquad
\hat S^z\simeq
\begin{pmatrix}
1&0&0\\
0&0&0\\
0&0&-1
\end{pmatrix},
$$

so

$$
\hat H_0\simeq
\begin{pmatrix}
1&0&0\\
0&0&0\\
0&0&1
\end{pmatrix},
\qquad
\hat V\simeq
\begin{pmatrix}
1&\frac{1}{\sqrt2}&0\\
\frac{1}{\sqrt2}&0&\frac{1}{\sqrt2}\\
0&\frac{1}{\sqrt2}&-1
\end{pmatrix}.
$$

Unperturbed manifolds: $E_1=1$ with $\{|+1\rangle,|-1\rangle\}$ (degenerate), and $E_0=0$ with $|0\rangle$.

**Step 1 (degenerate PT): construct effective blocks.**

For the $E_1$ manifold, to $O(\lambda^2)$,

$$
\hat H^{\text{eff}}_{1}\simeq
\begin{pmatrix}
1+\lambda+\frac{\lambda^2}{2} & \frac{\lambda^2}{2}\\
\frac{\lambda^2}{2} & 1-\lambda+\frac{\lambda^2}{2}
\end{pmatrix}
$$

(on the first-round corrected basis). For the $E_0$ manifold,

$$
\hat H^{\text{eff}}_{0}\simeq(-\lambda^2).
$$

At this stage, the original degeneracy is already lifted at $O(\lambda)$.

**Step 2 (second round: non-degenerate PT inside the lifted block).**

In $\hat H^{\text{eff}}_{1}$, the off-diagonal element is $\lambda^2/2$, while the diagonal splitting is

$$
\Delta E'=(1+\lambda+\lambda^2/2)-(1-\lambda+\lambda^2/2)=2\lambda.
$$

So the induced mixing scale is

$$
\frac{\lambda^2/2}{\Delta E'}=\frac{\lambda}{4}=O(\lambda).
$$

Hence the second-round **state** correction is $O(\lambda)$, but the induced **energy** correction is

$$
\frac{(\lambda^2/2)^2}{\Delta E'}\sim O(\lambda^3),
$$

so it does not modify results through $O(\lambda^2)$.

Therefore, up to $O(\lambda^2)$,

$$
\begin{aligned}
E_{+1}&=1+\lambda+\frac{\lambda^2}{2}+O(\lambda^3),\\
E_{-1}&=1-\lambda+\frac{\lambda^2}{2}+O(\lambda^3),\\
E_0&=-\lambda^2+O(\lambda^3).
\end{aligned}
$$

A representative first-order state structure is

$$
\begin{aligned}
|\pm1\rangle' &= |\pm1\rangle + \frac{\lambda}{\sqrt2}|0\rangle + O(\lambda^2),\\
|0\rangle' &= |0\rangle - \frac{\lambda}{\sqrt2}(|+1\rangle+|-1\rangle)+O(\lambda^2).
\end{aligned}
$$

**Interpretation.**
This example shows why degenerate perturbation theory is naturally hierarchical: first build and diagonalize effective blocks, then (if needed) iterate with non-degenerate perturbation inside the lifted blocks.
:::



:::{admonition} Discussion
:class: dropdown tip

If diagonalizing $V_{n\alpha,n\beta}$ still leaves repeated eigenvalues, first order is not enough. Then inspect second-order block terms (and symmetry constraints) to decide whether residual degeneracy is lifted.
:::

### Summary

- Degenerate perturbation theory is a block algorithm: solve each degenerate manifold before applying inter-manifold corrections.
- First-order splitting comes from diagonalizing $V_{n\alpha,n\beta}$ inside the manifold.
- Second-order shifts come from virtual transitions to outside manifolds and back.
- The same physics can be written in component form or projector form.
