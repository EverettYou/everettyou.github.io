## Lecture Notes

In Sec. 5.1.2, non-degenerate perturbation theory gave a clean algorithm for isolated levels. This subsection addresses the next step: when $\hat{H}_0$ has degenerate levels, how to reorganize the calculation so perturbation theory remains well-defined and computable.

### Overview

The key move in this subsection is to switch from **individual levels** to **degenerate subspaces**. We first diagnose why non-degenerate formulas diverge, then introduce projection operators to isolate one manifold, and finally derive the generalized Hellmann-Feynman relations and effective Hamiltonian used for computation.

The central outputs are: (i) first-order splitting from diagonalizing $\hat{P}_n\hat{V}\hat{P}_n$, (ii) second-order corrections from virtual transitions through $\hat{Q}_n$, and (iii) a step-by-step workflow for practical problems.

### Why non-degenerate formulas fail

For $\hat{H}(\lambda)=\hat{H}_0+\lambda\hat{V}$, suppose $\hat{H}_0$ has a $d$-fold degenerate eigenspace at energy $E_n$, spanned by $|n,\alpha\rangle$ ($\alpha=1,\dots,d$):

$$
\hat{H}_0|n,\alpha\rangle=E_n|n,\alpha\rangle.
$$

If we **naively** treat each $|n,\alpha\rangle$ as if it were an ordinary non-degenerate label, we would write

$$
|n,\alpha\rangle^{(1)}\stackrel{\text{naive}}{=}
\sum_{(m,\beta)\neq(n,\alpha)}
|m,\beta\rangle\frac{V_{m\beta,n\alpha}}{E_n-E_m},
\qquad
V_{m\beta,n\alpha}=\langle m,\beta|\hat V|n,\alpha\rangle.
$$

Now the problem is explicit: terms with $m=n$ but $\beta\neq\alpha$ are included in the sum, and for those terms the denominator is

$$
E_n-E_m=E_n-E_n=0.
$$

So the breakdown is not algebraic carelessness; it means the basis inside the degenerate subspace must be chosen first (by diagonalizing the in-subspace perturbation matrix).

:::{admonition} Core idea
:class: important

At degeneracy, first-order perturbation theory is a matrix problem in a subspace, not a scalar correction attached to each old level label.
:::

### Problem setup in a degenerate subspace

Use two indices to label the basis states:

$$
|n,\alpha\rangle,
$$

where $n$ labels the degenerate manifold and $\alpha$ labels orthonormal states within that manifold.

In this basis, write the unperturbed Hamiltonian and perturbation as

$$
\hat{H}_0=\sum_n\sum_\alpha E_n\,|n,\alpha\rangle\langle n,\alpha|,
$$

$$
\hat{V}=\sum_{m,\alpha}\sum_{n,\beta}|m,\alpha\rangle V_{m\alpha,n\beta}\langle n,\beta|,
\qquad
V_{m\alpha,n\beta}=\langle m,\alpha|\hat{V}|n,\beta\rangle.
$$

Then the full Hamiltonian is simply

$$
\hat{H}(\lambda)=\hat{H}_0+\lambda\hat{V}.
$$

Because states inside one manifold can mix, the corresponding eigenvalue problem is matrix-valued in $\alpha,\beta$:

$$
\hat{H}(\lambda)|n,\beta(\lambda)\rangle
=\sum_\alpha E_{n,\alpha\beta}(\lambda)\,|n,\alpha(\lambda)\rangle.
$$

### Generalized Hellmann-Feynman theorem

Project and differentiate at $\lambda=0$:

$$
\langle m,\gamma|\partial_\lambda\hat{H}|n,\beta\rangle
=\partial_\lambda E_{n,\gamma\beta}\,\delta_{mn}
+(E_n-E_m)\langle m,\gamma|\partial_\lambda n,\beta\rangle.
$$

So the two key statements are:

- **within the same manifold** $(m=n)$,

$$
\partial_\lambda E_{n,\alpha\beta}=V_{n\alpha,n\beta};
$$

- **between different manifolds** $(m\neq n)$,

$$
\langle m,\alpha|\partial_\lambda n,\beta\rangle
=\frac{V_{m\alpha,n\beta}}{E_n-E_m}.
$$

Interpretation: first equation gives in-subspace splitting; second equation gives mixing with outside manifolds.

:::{admonition} Derivation sketch
:class: dropdown information

Differentiate
$\hat{H}|n,\beta\rangle=\sum_\alpha |n,\alpha\rangle E_{n,\alpha\beta}$,
left-multiply by $\langle m,\gamma|$, and use
$\langle m,\gamma|\hat{H}_0=E_m\langle m,\gamma|$.
Then separate $m=n$ and $m\neq n$.
:::

### Energy corrections in component form

Using the same derivative expansion strategy as Sec. 5.1.2, the effective energy matrix in the $n$th degenerate manifold is

$$
E_{n,\alpha\beta}(\lambda)
=E_n\delta_{\alpha\beta}
+\lambda V_{n\alpha,n\beta}
+\lambda^2\sum_{m\neq n}\sum_\gamma
\frac{V_{n\alpha,m\gamma}V_{m\gamma,n\beta}}{E_n-E_m}
+O(\lambda^3).
$$

Interpretation:

- the $O(\lambda)$ term is in-subspace splitting;
- the $O(\lambda^2)$ term is virtual coupling through outside manifolds.

### State corrections in component form

The corrected basis states also receive perturbative corrections. To first order,

$$
|n,\alpha(\lambda)\rangle
=|n,\alpha\rangle
+\lambda\sum_{m\neq n}\sum_{\beta}
|m,\beta\rangle\frac{V_{m\beta,n\alpha}}{E_n-E_m}
+O(\lambda^2).
$$

This is the degenerate analog of the non-degenerate state-mixing formula: states in the target manifold hybridize with outside manifolds according to coupling strength over energy gap.

:::{admonition} Note on higher-order state corrections
:class: dropdown information

Second-order state corrections can also be derived in component form, but they are algebraically lengthy. For most practical calculations in this chapter, the key steps are:

1. diagonalize the in-manifold first-order matrix,
2. add second-order energy-matrix corrections,
3. use first-order state mixing to track how basis states are dressed.
:::

:::{admonition} Optional compact operator form
:class: dropdown information

The same energy correction result can be written with projectors
$\hat{P}_n=\sum_\alpha |n,\alpha\rangle\langle n,\alpha|$ and
$\hat{Q}_n=\hat{I}-\hat{P}_n$ as

$$
\hat{H}^{\text{eff}}_n
=E_n\hat{I}_d
+\lambda\hat{P}_n\hat{V}\hat{P}_n
+\lambda^2\hat{P}_n\hat{V}\hat{Q}_n
\frac{1}{E_n-\hat{Q}_n\hat{H}_0\hat{Q}_n}
\hat{Q}_n\hat{V}\hat{P}_n
+O(\lambda^3).
$$

In this lecture, we compute mainly in component form.
:::

:::{admonition} Example 1: Spin-1 model
:class: example

**Why this example?**
This is the minimal finite-dimensional model where degeneracy appears naturally and is lifted in a controlled way. It lets us practice the full workflow (identify subspace $\to$ build block matrix $\to$ compute corrections) with very transparent matrix elements.

**Problem setup.**

$$
\hat{H}_0=(\hat{S}^z)^2,
\qquad
\hat{V}=\lambda(\hat{S}^x+\hat{S}^z).
$$

We treat $\lambda$ as a small parameter and compute the perturbed spectrum order by order.

**Step 1. Identify degenerate manifold.**

$E=1$ manifold: $\{|+1\rangle,|-1\rangle\}$; $E=0$ manifold: $|0\rangle$.

**Step 2. Build first-order matrix inside $E=1$ manifold.**

$$
V_{\alpha\beta}=\langle\alpha|\hat{V}|\beta\rangle,
\quad
\alpha,\beta\in\{+1,-1\}.
$$

Because $\hat{S}^x$ connects only $\Delta m=\pm1$, the off-diagonal elements vanish:

$$
\langle+1|\hat{S}^x|-1\rangle=\langle-1|\hat{S}^x|+1\rangle=0.
$$

And

$$
\langle\pm1|\hat{S}^z|\pm1\rangle=\pm1.
$$

Hence

$$
V_{\alpha\beta}=\lambda
\begin{pmatrix}
1&0\\0&-1
\end{pmatrix},
$$

so first-order splittings are

$$
\Delta E^{(1)}_{\pm1}=\pm\lambda.
$$

**Step 3. Compute second-order shifts from outside states.**

Only outside state is $|0\rangle$, with

$$
\langle\pm1|\hat{V}|0\rangle
=\lambda\langle\pm1|\hat{S}^x|0\rangle
=\frac{\lambda}{\sqrt2}.
$$

Therefore

$$
E_{+1}^{(2)}=\frac{|\lambda/\sqrt2|^2}{1-0}=\frac{\lambda^2}{2},
\qquad
E_{-1}^{(2)}=\frac{|\lambda/\sqrt2|^2}{1-0}=\frac{\lambda^2}{2},
$$

$$
E_0^{(2)}=\frac{|\lambda/\sqrt2|^2}{0-1}+\frac{|\lambda/\sqrt2|^2}{0-1}=-\lambda^2.
$$

Final result (to second order):

$$
E_{\pm1}\approx1\pm\lambda+\frac{\lambda^2}{2},
\qquad
E_0\approx-\lambda^2.
$$
:::

:::{admonition} Example 2: Three-level dark-state model
:class: example

**Why this example?**
This model gives a clear physical picture of second-order degenerate perturbation theory: two low-energy states are coupled only through a virtual high-energy state, producing bright/dark combinations. It is a standard mechanism behind effective interactions in atomic, optical, and condensed-matter settings.

**Problem setup.**

$$
\hat{H}_0=\Delta|3\rangle\langle3|,
\qquad
\hat{V}=\lambda_1|1\rangle\langle3|+\lambda_2|2\rangle\langle3|+h.c.
$$

with $|\lambda_{1,2}|\ll\Delta$.

**Step 1. Identify degenerate manifold and outside states.**

Degenerate manifold: $\{|1\rangle,|2\rangle\}$ at energy $0$.
Outside state: $|3\rangle$ at energy $\Delta$.

**Step 2. First-order matrix inside manifold.**

$$
\langle i|\hat{V}|j\rangle=0,
\quad i,j\in\{1,2\},
$$

so there is no first-order splitting.

**Step 3. Second-order effective matrix elements.**

$$
(H^{\text{eff}})_{ij}^{(2)}
=\sum_{m\notin\{1,2\}}
\frac{\langle i|\hat{V}|m\rangle\langle m|\hat{V}|j\rangle}{0-E_m}
= -\frac{1}{\Delta}\langle i|\hat{V}|3\rangle\langle3|\hat{V}|j\rangle.
$$

Thus

$$
\hat{H}^{\text{eff}}=-\frac{1}{\Delta}
\begin{pmatrix}
|\lambda_1|^2 & \lambda_1\lambda_2^*\\
\lambda_2\lambda_1^* & |\lambda_2|^2
\end{pmatrix}.
$$

**Step 4. Diagonalize and identify bright/dark modes.**

One eigenvalue is

$$
E_B=-\frac{|\lambda_1|^2+|\lambda_2|^2}{\Delta},
$$

with bright vector proportional to
$\lambda_1|1\rangle+\lambda_2|2\rangle$.

The orthogonal combination

$$
|D\rangle=
\frac{\lambda_2^*|1\rangle-\lambda_1^*|2\rangle}{\sqrt{|\lambda_1|^2+|\lambda_2|^2}}
$$

has eigenvalue $0$.

**Step 5. Exact check.**

Directly, $\hat{H}_0|D\rangle=0$ and $\hat{V}|D\rangle=0$, so $|D\rangle$ is an exact zero mode of $\hat{H}_0+\hat{V}$.
:::

:::{admonition} Discussion
:class: dropdown tip

If the first-order matrix $V_{n\alpha,n\beta}$ still has repeated eigenvalues, first-order theory leaves residual degeneracy. Then second-order block terms (and symmetry) determine whether that residual degeneracy is lifted.
:::

### Summary

- Degenerate perturbation theory is a block-level method: solve each degenerate manifold first.
- First-order splitting is obtained by diagonalizing the matrix $V_{n\alpha,n\beta}=\langle n,\alpha|\hat{V}|n,\beta\rangle$.
- Second-order corrections come from virtual transitions to outside manifolds and back.
- Component form and projector form are equivalent; use either as long as matrix elements are computed carefully.
