## Lecture Notes

### Overview

This subsection answers one concrete question: **what to do when the unperturbed spectrum is degenerate**. The teaching strategy follows the same logic as the toy model in 5.1.1 and the chapter reference notes:

1. identify exactly where non-degenerate perturbation theory breaks,
2. reorganize the problem into degenerate subspaces,
3. solve each subspace with an effective Hamiltonian,
4. then continue with ordinary perturbation theory after degeneracy is lifted.

### Why Non-Degenerate Formulas Break

For a degenerate level $E_n^{(0)}$, choose basis states $\vert n,\alpha\rangle$ with $\alpha=1,\dots,d$.

If we directly use the non-degenerate first-order state correction,

$$
\vert\psi_n^{(1)}\rangle = \sum_{m\neq n}
\frac{\langle m^{(0)}\vert V\vert n^{(0)}\rangle}{E_n^{(0)}-E_m^{(0)}}
\vert m^{(0)}\rangle,
$$

then terms with $m$ inside the same degenerate manifold give $E_n^{(0)}-E_m^{(0)}=0$. The divergence is not a technical accident; it means the zeroth-order basis is not yet the right one.

:::{admonition} Core Fix
:class: important

Do **not** correct states level-by-level first. Instead, treat each degenerate manifold as one block, and diagonalize $V$ inside that block.
:::

### Divide-and-Conquer View (Degenerate Subspaces)

Introduce projectors

$$
P_n=\sum_{\alpha=1}^d\vert n,\alpha\rangle\langle n,\alpha\vert,
\qquad
Q_n=\hat I-P_n.
$$

Inside one manifold, the perturbation is the matrix

$$
W^{(n)}_{\alpha\beta}=\langle n,\alpha\vert V\vert n,\beta\rangle.
$$

- Eigenvectors of $W^{(n)}$ give the **good zeroth-order basis** inside the degenerate subspace.
- Eigenvalues of $W^{(n)}$ give **first-order energy splitting**.

This is the degenerate counterpart of Hellmann-Feynman: first-order physics is matrix-valued within a degenerate block, not scalar-valued level-by-level.

### Effective Hamiltonian to Second Order

The effective Hamiltonian in the $n$th degenerate manifold is

$$
H_{n}^{\mathrm{eff}}
=E_n^{(0)}I_d
+\lambda P_nVP_n
+\lambda^2 P_nVQ_n\frac{1}{E_n^{(0)}-Q_nH_0Q_n}Q_nVP_n
+O(\lambda^3).
$$

Interpretation:

- first term: unperturbed degenerate energy,
- first-order term: direct mixing/splitting inside the manifold,
- second-order term: virtual excursions to outside states and return.

After diagonalizing $H_n^{\mathrm{eff}}$ (to the order you keep), you can continue with non-degenerate perturbation theory for separated levels.

### Application 1: Spin-1 Model (from Chapter Reference)

Take

$$
H_0=(S^z)^2,
\qquad
V=\lambda(S^x+S^z).
$$

Unperturbed levels:

- $E^{(0)}=1$: two-fold degenerate $\{|+1\rangle,|-1\rangle\}$,
- $E^{(0)}=0$: non-degenerate $|0\rangle$.

For the degenerate block $\{|+1\rangle,|-1\rangle\}$,

$$
P_1VP_1
=\lambda
\begin{pmatrix}
\langle +1|S^z|+1\rangle & \langle +1|S^x+S^z|-1\rangle\\
\langle -1|S^x+S^z|+1\rangle & \langle -1|S^z|-1\rangle
\end{pmatrix}
=\lambda
\begin{pmatrix}
1&0\\0&-1
\end{pmatrix}.
$$

So degeneracy is already lifted at first order:

$$
\Delta E_{\pm1}^{(1)}=\pm\lambda.
$$

Second-order virtual mixing via $|0\rangle$ gives

$$
E_{\pm1}\approx 1\pm\lambda+\frac{\lambda^2}{2},
\qquad
E_0\approx-\lambda^2.
$$

### Application 2: Three-Level Dark-State Toy Problem

Reference setup:

$$
H_0=\Delta|3\rangle\langle3|,
\qquad
V=\lambda_1|1\rangle\langle3|+\lambda_2|2\rangle\langle3|+h.c.
$$

The degenerate manifold is $\{|1\rangle,|2\rangle\}$ at zero energy. To second order,

$$
H_{\mathrm{eff}}=-\frac{1}{\Delta}
\begin{pmatrix}
|\lambda_1|^2 & \lambda_1\lambda_2^*\\
\lambda_2\lambda_1^* & |\lambda_2|^2
\end{pmatrix}
=-\frac{1}{\Delta}|\lambda\rangle\langle\lambda|,
$$

with $|\lambda\rangle=\lambda_1|1\rangle+\lambda_2|2\rangle$.

Hence one bright mode shifts and one dark mode stays at zero:

$$
E_+=-\frac{|\lambda_1|^2+|\lambda_2|^2}{\Delta},
\qquad E_-=0.
$$

A normalized dark state is

$$
|D\rangle
=\frac{\lambda_2^*|1\rangle-\lambda_1^*|2\rangle}
{\sqrt{|\lambda_1|^2+|\lambda_2|^2}},
$$

and is an exact zero mode of $H_0+V$.

:::{admonition} Discussion
:class: dropdown tip

If $P_nVP_n$ still has repeated eigenvalues, first-order theory leaves residual degeneracy. Then second-order terms (or symmetry analysis) decide whether and how that residual degeneracy is lifted.
:::

### Summary

- Degenerate perturbation theory is a **block perturbation** method: solve each degenerate manifold first, not each level first.
- First-order physics comes from diagonalizing $P_nVP_n$ inside the manifold.
- Second-order terms describe virtual transitions through $Q_n$ states and generate additional splitting/shifts.
- The workflow is hierarchical: block diagonalization first, ordinary perturbation theory next.
